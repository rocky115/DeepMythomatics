# Recurrence and Transience in Markov Chains

In the study of Markov chains, the notions of recurrence and transience play a crucial role in understanding the behavior of the chain over time. Below, we explore these concepts in detail, including their formal definitions and examples.

## Weak Stability in Markov Chains

While stability in a Markov chain $ X $ is desirable, it may not be sufficient to ensure that the trajectory of $ X $ enters a set $ A $ frequently enough. For instance, consider a maximization problem where a random walk explores the surface of a function to maximize its value. Without a systematic sweep of this surface, convergence to the global maximum cannot be guaranteed. This need for systematic behavior introduces different notions of recurrence.

## Recurrence in Discrete State Spaces

In a discrete setup, recurrence ensures that a state will be revisited with certainty. For irreducible chains in a finite state space, recurrence or transience becomes a property of the chain itself, not of individual states.

### Definition 6.25: Recurrence and Transience

Let $ X $ be a finite state space. A state $ w \in X $ is:

- **Transient** if the average number of visits to $ w $, $ \mathbb{E}[n_w] $, is finite.  
- **Recurrent** if $ \mathbb{E}[n_w] = \infty $.

For irreducible chains, recurrence and transience are properties of the entire chain. From the Chapman-Kolmogorov equations, for every $ (x, y) \in X^2 $:

- $ \mathbb{E}_x[n_y] < \infty $ in the transient case.
- $ \mathbb{E}_x[n_y] = \infty $ in the recurrent case.

Thus, an irreducible chain is classified as either transient or recurrent.

---

## Example 26: Branching Process

Consider a population model where individuals reproduce independently. Let $ X $ represent the number of siblings of an individual, distributed according to a generating function $ g(s) = \mathbb{E}[s^X] $. If reproduction occurs in fixed generations, the size of the $ t $-th generation $ S_t $ (for $ t \geq 1 $) is given by:

$$
S_t = \sum_{i=1}^{S_{t-1}} X_i,
$$

where the $ X_i $ are independent random variables. Starting with a single individual at $ t = 0 $, we have $ S_0 = 1 $, and the generating function of $ S_t $ is $ g_t(s) = g(g_{t-1}(s)) $, with $ g_0(s) = s $.

### Transience and Recurrence in the Branching Process

1. **Case 1: Transience**  
   If $ P(X = 0) = 0 $, the chain $ (S_t) $ is transient because the population size is strictly increasing.

2. **Case 2: Recurrence**  
   If $ P(X = 0) > 0 $, there is a probability $ p_t = P(S_t = 0) = g_t(0) $. This satisfies the recurrence equation:

   $$
   p_t = g(p_{t-1}).
   $$

   A limiting probability $ p $ exists such that $ p = g(p) $, provided $ g'(1) = \mathbb{E}[X] \leq 1 $. The chain is recurrent if $ \mathbb{E}[X] \leq 1 $.

3. **Geometric Distribution of Returns**  
   If a restarting mechanism resets $ S_t $ to 0 upon extinction, and $ g'(1) > 1 $, the number of returns to 0 follows a geometric distribution with parameter $ p $.

---

## Extending Recurrence and Transience to General Cases

In general state spaces, the notions of recurrence and transience extend to chains with atoms. For non-discrete cases, these extensions rely on small sets and splitting techniques, as outlined by Athreya and Ney (1978).

### Key Insights

- **Irreducible Chains:** In finite state spaces, every irreducible chain is either transient or recurrent.
- **Branching Process Behavior:** Whether a branching process is transient or recurrent depends on the average number of siblings per individual.

The study of recurrence and transience provides a foundation for understanding the long-term behavior of Markov chains in both discrete and general settings.

                                                                          
  # Recurrence and Transience of Sets in Markov Chains

In this section, we explore the properties of recurrence and transience as applied to sets in Markov chains. The definitions and theorems presented provide a deeper understanding of how these properties influence the behavior of Markov chains.

---

## Definition 27: Recurrent and Transient Sets

1. **Recurrent Set**  
   A set $ A \subseteq X $ is called **recurrent** if:  
   $$
   \mathbb{E}_x[N_A] = +\infty \quad \text{for every } x \in A,
   $$  
   where $ N_A $ is the number of visits to $ A $.

2. **Uniformly Transient Set**  
   A set $ A $ is **uniformly transient** if there exists a constant $ M $ such that:  
   $$
   \mathbb{E}_z[N_A] < M \quad \text{for every } z \in A.
   $$

3. **Transient Set**  
   A set $ A $ is **transient** if there exists a covering of $ X $ by uniformly transient sets. That is, there exists a countable collection of uniformly transient sets $ \{B_i\} $ such that:  
   $$
   A = \bigcup_{i} B_i.
   $$

---

## Theorem 28: Recurrence and Transience in $ \psi $-Irreducible Chains

Let $ (X_n) $ be a $ \psi $-irreducible Markov chain with an accessible atom $ a $:

1. **Recurrent Atom**  
   If $ a $ is recurrent, every set $ A $ in $ \mathcal{B}(X) $ such that $ \psi(A) > 0 $ is recurrent.

2. **Transient Atom**  
   If $ a $ is transient, then $ X $ is transient.

**Notes**:
- Property (i) is critical in Markov Chain Monte Carlo (MCMC) contexts and can be derived using the Chapman-Kolmogorov equations.
- Property (ii) is more challenging to establish, as it relies on $ P_a(T_a < \infty) < 1 $ for a transient atom, with $ T_a $ being the time of the last visit to $ a $ (see Meyn and Tweedie 1993, p. 181).

---

## Definition 29: Recurrence and Transience of a Markov Chain

1. **Recurrent Chain**  
   A Markov chain $ (X_n) $ is **recurrent** if:
   - There exists a measure $ \psi $ such that $ (X_n) $ is irreducible.
   - For every $ A \in \mathcal{B}(X) $ with $ \psi(A) > 0 $:  
     $$
     \mathbb{E}_z[N_A] = \infty \quad \text{for every } z \in A.
     $$

2. **Transient Chain**  
   A Markov chain $ (X_n) $ is **transient** if it is irreducible and every state $ x \in X $ is transient.

---

## Theorem. 30: Classification of Irreducible Chains

An irreducible Markov chain is either recurrent or transient.

**Proof Outline**:
- Strongly aperiodic chains satisfy a minorizing condition, allowing them to be split into components as in Equation (6.3.2).
- For an irreducible chain $ (X_n) $ and its split version $ (X_n') $, recurrence or transience is preserved.
- The classification extends to general irreducible chains through their corresponding $ \psi $-chains, which are strongly aperiodic. 

Using the relation:  
$$
\mathbb{E}_z[N_A] = \sum_{n=0}^\infty K^n(z, A) = \sum_{n=0}^\infty K_\psi^n(z, A),
$$  
where $ K_\psi $ is the transition kernel for the $ \psi $-chain, the recurrence or transience of the chain can be determined.

---

This framework establishes the foundational properties of recurrence and transience, enabling further analysis of Markov chains in both theoretical and applied contexts.

# Criteria for Recurrence in Markov Chains

The dichotomy between recurrence and transience for irreducible Markov chains provides a foundation for analyzing their behavior. To simplify the verification of recurrence, alternative criteria can be developed, particularly using small sets and related conditions.

---

## ** Criteria for Recurrence**

### Proposition 6.31: Recurrence Based on Small Sets

An irreducible Markov chain $ (X_n) $ is **recurrent** if there exists a small set $ C $ with $ \pi(C) > 0 $ such that:

$$
P_a(T_C < \infty) = 1 \quad \text{for every } a \in C,
$$

where $ T_C $ is the first hitting time of $ C $.

#### **Proof:**

1. **Recurrent Atom $ C $:**  
   For $ C \subseteq X $, define:
   - $ U_n = K^n(x, C) $, the probability of visiting $ C $ at time $ n $.
   - $ f_n = P(X_n \in C, X_{n-1} \notin C, \dots, X_1 \notin C) $, the probability of the first visit to $ C $ at time $ n $.

   Define:
   $$
   U(s) = \sum_{n=1}^\infty U_n s^n, \quad Q(s) = \sum_{n=1}^\infty f_n s^n.
   $$

   The relationship between $ U_n $ and $ f_n $ is given by:

   $$
   U_n = f_n + \sum_{m=1}^{n-1} f_m U_{n-m}.
   $$

   Transforming into generating functions, this becomes:
   $$
   U(s) = \frac{1}{1 - Q(s)}.
   $$

2. **Connection Between Recurrence and $ Q(1) $:**  
   - $ Q(1) = P(T_C < \infty) = 1 $, which implies that $ \mathbb{E}_x[N_C] = \infty $ (recurrence of $ C $).

3. **Extension to Split Chain:**  
   The proof extends to the split chain $ (X_n) $ by showing that visits to $ C \times \{0\} $ ensure independence by renewal. Using $ \mathbb{E}[N_C] $ for the original chain and its split version, recurrence is established for $ (X_n) $ by Theorem 6.28, since $ C \times \{0\} $ is a recurrent atom.

---

### Alternative Criterion: Generalized Drift Condition

Another approach to verifying recurrence generalizes the notions of small sets and minorizing conditions. This involves:

1. **Potential Function $ V(x) $:** A non-negative function.
2. **Drift Condition:** A condition like:
   $$
   \Delta V(x) \leq -f(x) + b \mathbf{1}_C(x),
   $$
   where $ f(x) > 0 $ ensures a "downward drift" outside of a small set $ C $.

This criterion relies on the transition kernel $ K(\cdot, \cdot) $ rather than powers $ K^n $, with detailed implications for stability and convergence discussed in **Note 6.9.1**.

---

## ** Harris Recurrence**

Harris strengthened the concept of recurrence by considering the probability of visiting a set $ A $ an infinite number of times across all paths.

### Harris Recurrence

A Markov chain $ (X_n) $ is **Harris recurrent** if for every measurable set $ A $ with $ \pi(A) > 0 $:

$$
P_x(\eta_A = \infty) = 1 \quad \text{for every } x \in X,
$$

where $ \eta_A $ is the number of visits to $ A $.

#### Key Features:
1. **Infinite Visits on Every Path:** Harris recurrence guarantees that every sample path visits $ A $ infinitely often.
2. **Stronger Stability:** This notion extends the stability properties of ordinary recurrence by requiring recurrence across all paths, not just in expectation.

Harris recurrence is particularly significant in stability and convergence studies of Markov chains.

```python
import numpy as np

class MarkovChain:
    def __init__(self, transition_matrix):
        """
        Initialize a Markov chain with the given transition matrix.
        
        Args:
            transition_matrix (numpy.ndarray): Transition matrix of the Markov chain.
        """
        self.P = np.array(transition_matrix)
        self.num_states = self.P.shape[0]
    
    def is_irreducible(self):
        """
        Check if the Markov chain is irreducible.
        """
        reachability_matrix = np.linalg.matrix_power(self.P > 0, self.num_states)
        return np.all(reachability_matrix > 0)

    def hitting_probability(self, start_state, target_state, max_steps=100):
        """
        Compute the probability of hitting the target state from the start state.
        
        Args:
            start_state (int): Index of the starting state.
            target_state (int): Index of the target state.
            max_steps (int): Maximum steps to consider for hitting probability.

        Returns:
            float: Probability of reaching the target state.
        """
        hitting_prob = 0
        current_prob = np.eye(self.num_states)
        for _ in range(max_steps):
            current_prob = np.dot(current_prob, self.P)
            hitting_prob += current_prob[start_state, target_state]
            if hitting_prob >= 1 - 1e-10:  # Close to 1
                break
        return hitting_prob
def is_recurrent(chain, state, max_steps=100):
    """
    Check if a given state in the Markov chain is recurrent.
    
    Args:
        chain (MarkovChain): An instance of the MarkovChain class.
        state (int): Index of the state to check.
        max_steps (int): Maximum steps to compute recurrence.

    Returns:
        bool: True if the state is recurrent, False otherwise.
    """
    prob = chain.hitting_probability(state, state, max_steps=max_steps)
    return prob >= 1.0

def is_harris_recurrent(chain, small_set_indices, max_steps=100):
    """
    Check if the Markov chain satisfies Harris recurrence for a given small set.
    
    Args:
        chain (MarkovChain): An instance of the MarkovChain class.
        small_set_indices (list): List of state indices defining the small set.
        max_steps (int): Maximum steps to compute recurrence.

    Returns:
        bool: True if the chain is Harris recurrent, False otherwise.
    """
    for state in small_set_indices:
        if not is_recurrent(chain, state, max_steps=max_steps):
            return False
    return True

# Example: Transition matrix for a simple Markov chain
P = [
    [0.5, 0.5, 0.0],
    [0.2, 0.3, 0.5],
    [0.0, 0.1, 0.9]
]

chain = MarkovChain(P)

# Check irreducibility
print("Is the chain irreducible?", chain.is_irreducible())

# Check if a state is recurrent
state = 0
print(f"Is state {state} recurrent?", is_recurrent(chain, state))

# Harris recurrence with a small set {0, 1}
small_set = [0, 1]
print("Is the chain Harris recurrent for the small set?", is_harris_recurrent(chain, small_set))

import networkx as nx
import matplotlib.pyplot as plt

def visualize_markov_chain(P):
    """
    Visualize the Markov chain as a directed graph.
    
    Args:
        P (numpy.ndarray): Transition matrix of the Markov chain.
    """
    num_states = P.shape[0]
    G = nx.DiGraph()

    for i in range(num_states):
        for j in range(num_states):
            if P[i, j] > 0:
                G.add_edge(i, j, weight=P[i, j])

    pos = nx.spring_layout(G)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=10)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={k: f"{v:.2f}" for k, v in labels.items()})
    plt.title("Markov Chain Transition Graph")
    plt.show()

# Visualize the example Markov chain
visualize_markov_chain(np.array(P))


```

```python
# Markov Chain Implementation
class MarkovChain:
    def __init__(self, transition_matrix):
        """
        Initialize a Markov chain with the given transition matrix.

        Args:
            transition_matrix (list of lists): Transition matrix of the Markov chain.
        """
        self.P = transition_matrix
        self.num_states = len(transition_matrix)

    def is_irreducible(self):
        """
        Check if the Markov chain is irreducible.

        Returns:
            bool: True if irreducible, False otherwise.
        """
        reachability = [[False] * self.num_states for _ in range(self.num_states)]

        # Initialize reachability matrix
        for i in range(self.num_states):
            for j in range(self.num_states):
                reachability[i][j] = self.P[i][j] > 0

        # Perform reachability via matrix multiplication
        for _ in range(self.num_states - 1):
            new_reachability = [[False] * self.num_states for _ in range(self.num_states)]
            for i in range(self.num_states):
                for j in range(self.num_states):
                    for k in range(self.num_states):
                        new_reachability[i][j] |= reachability[i][k] and reachability[k][j]
            reachability = new_reachability

        # Check if all states are mutually reachable
        for i in range(self.num_states):
            for j in range(self.num_states):
                if not reachability[i][j]:
                    return False
        return True

    def hitting_probability(self, start_state, target_state, max_steps=100):
        """
        Compute the probability of hitting the target state from the start state.

        Args:
            start_state (int): Index of the starting state.
            target_state (int): Index of the target state.
            max_steps (int): Maximum steps to consider for hitting probability.

        Returns:
            float: Probability of reaching the target state.
        """
        current_prob = [[1 if i == j else 0 for j in range(self.num_states)] for i in range(self.num_states)]
        hitting_prob = 0

        for _ in range(max_steps):
            # Multiply the probability matrix with the transition matrix
            next_prob = [[0] * self.num_states for _ in range(self.num_states)]
            for i in range(self.num_states):
                for j in range(self.num_states):
                    for k in range(self.num_states):
                        next_prob[i][j] += current_prob[i][k] * self.P[k][j]

            hitting_prob += next_prob[start_state][target_state]
            current_prob = next_prob

            # Stop early if hitting probability is close to 1
            if hitting_prob >= 1 - 1e-10:
                break

        return hitting_prob

# Recurrence and Harris Recurrence
def is_recurrent(chain, state, max_steps=100):
    """
    Check if a given state in the Markov chain is recurrent.

    Args:
        chain (MarkovChain): Instance of the MarkovChain class.
        state (int): Index of the state to check.
        max_steps (int): Maximum steps to compute recurrence.

    Returns:
        bool: True if the state is recurrent, False otherwise.
    """
    prob = chain.hitting_probability(state, state, max_steps=max_steps)
    return prob >= 1.0

def is_harris_recurrent(chain, small_set_indices, max_steps=100):
    """
    Check if the Markov chain satisfies Harris recurrence for a given small set.

    Args:
        chain (MarkovChain): Instance of the MarkovChain class.
        small_set_indices (list): List of state indices defining the small set.
        max_steps (int): Maximum steps to compute recurrence.

    Returns:
        bool: True if the chain is Harris recurrent, False otherwise.
    """
    for state in small_set_indices:
        if not is_recurrent(chain, state, max_steps=max_steps):
            return False
    return True

# Example Usage
if __name__ == "__main__":
    # Example transition matrix
    P = [
        [0.5, 0.5, 0.0],
        [0.2, 0.3, 0.5],
        [0.0, 0.1, 0.9]
    ]

    chain = MarkovChain(P)

    # Check irreducibility
    print("Is the chain irreducible?", chain.is_irreducible())

    # Check if a state is recurrent
    state = 0
    print(f"Is state {state} recurrent?", is_recurrent(chain, state))

    # Harris recurrence with a small set {0, 1}
    small_set = [0, 1]
    print("Is the chain Harris recurrent for the small set?", is_harris_recurrent(chain, small_set))
```

## Definition 32: Harris Recurrence

A set $ A $ is **Harris recurrent** if $ P(A = \infty) = 1 $ for all $ x \in A $. A Markov chain $ (X_n) $ is **Harris recurrent** if there exists a measure $ \phi $ such that $ (X_n) $ is $ \phi $-irreducible and, for every set $ A $ with $ \phi(A) > 0 $, $ A $ is Harris recurrent.

Recurrence corresponds to $ E_z[\eta_A] = \infty $, which is a weaker condition than $ P(A = \infty) = 1 $. The following proposition expresses Harris recurrence as a condition on $ P(T_A < \infty) $, where $ T_A $ is the first hitting time defined as:

$$
T_A = \inf \{ n \geq 1 : X_n \in A \}.
$$

### Proposition 6.33: Harris Recurrence Condition

If for every $ A \in \mathcal{B}(X) $, $ P_x(T_A < \infty) = 1 $ for every $ x \in A $, then $ P(A = \infty) = 1 $ for all $ x \in X $, and $ (X_n) $ is Harris recurrent.

#### Proof

The average number of visits to $ B $ before the first visit to $ A $ is given by:
$$
U_A(x, B) = \sum_{n=1}^\infty P_x(X_n \in B, T_A \geq n).
$$

From this, $ U_A(x, A) = P_x(T_A < \infty) $, since if $ B \subseteq A $:
$$
P_x(X_n \in B, T_A \geq n) = P_x(X_n \in B, T_A = n) = P_x(T_B = n).
$$

If $ T_A(k) $ (for $ k > 1 $) denotes the time of the $ k $-th visit to $ A $, $ T_A(k) $ satisfies:

$$
P_x(T_A(k+1) < \infty) = P_x(T_A(k) < \infty) \int_A U_A(y, A) \phi(dy) = P_x(T_A(k) < \infty).
$$

By induction:
$$
P_x(A > k) = P_x(T_A(k) < \infty).
$$

Taking the limit as $ k \to \infty $, $ P_x(A = \infty) = 1 $ for $ x \in A $.

---

### Theorem 34: Sufficient Condition for Harris Recurrence

If $ (X_n) $ is a $ \phi $-irreducible Markov chain with a small set $ C $ such that $ P_x(T_C < \infty) = 1 $ for all $ x \in X $, then $ (X_n) $ is Harris recurrent.

#### Notes
- The property of Harris recurrence is primarily needed when $ X $ is not countable.
- If $ X $ is finite or countable, $ E_z[\eta_A] = \infty $ if and only if $ P_x(T_A < \infty) = 1 $, as shown in the proof of Proposition 6.31.
- Harris recurrence provides a sufficient condition for strong stability and convergence properties of the Markov chain.


##  Invariant Measures

###  Stationary Chains

An increased level of stability for the Markov chain $ (X_n) $ is achieved if the marginal distribution of $ X_n $ is independent of $ n $. This condition ensures the existence of a probability distribution under which $ X_n \sim \pi $ for all $ n $. Markov chain Monte Carlo (MCMC) methods rely on this property, which is associated with **positive recurrence**.

The Markov chains constructed from MCMC algorithms typically exhibit this enhanced stability property, except in highly pathological cases (see Section 10.4.3). Below is an introduction to invariant measures and their relationship with positive recurrence.

---

### Definition 35: Invariant Measure

A $ \sigma $-finite measure $ \pi $ is **invariant** for the transition kernel $ K(x, \cdot) $ (and the associated chain $ (X_n) $) if:

$$
\pi(B) = \int_X K(x, B) \pi(dx), \quad \forall B \in \mathcal{B}(X).
$$

When an invariant probability measure exists for a $ \phi $-irreducible chain (which is recurrent by Theorem 6.30), the chain is called **positive recurrent**. Chains that are recurrent but do not admit a finite invariant measure are referred to as **null recurrent**.

The invariant distribution is also referred to as **stationary** if $ \pi $ is a probability measure, since $ X_0 \sim \pi $ implies $ X_n \sim \pi $ for every $ n $. Thus, the chain is stationary in distribution.

---

### Uniqueness of the Invariant Measure

For an irreducible chain that allows a $ \sigma $-finite invariant measure, this measure is unique up to a multiplicative constant. (See Problem 6.60 for details.) 

---

### Proposition 36: Positivity and Recurrence

If the chain $ (X_n) $ is **positive recurrent**, it is also recurrent.

#### Discussion

This proposition formalizes the intuition that the existence of an invariant measure prevents the probability mass from "escaping to infinity." The invariant measure provides a stable distribution for the chain, ensuring that it revisits any set with positive measure infinitely often. 

---

### Harris Recurrence in MCMC Algorithms

Chan and Geyer (1994) and Tierney (1994) highlight the importance of Harris recurrence in the context of MCMC algorithms. Harris recurrence ensures that the asymptotic properties of the chain are robust and do not depend on the starting distribution. The **split chain construction** used in this context eliminates measure-theoretic pathologies and guarantees convergence to the invariant distribution under mild assumptions.

##  Kac's Theorem

A classical result in the theory of irreducible Markov chains with a discrete state space is that the stationary distribution, when it exists, can be expressed in terms of the expected return time to a state. This result, often referred to as **Kac's Theorem**, is given by:

$$
\pi(x) = \frac{1}{\mathbb{E}_x[T_x]},
$$

where $ \mathbb{E}_x[T_x] $ is the expected number of steps between successive visits to $ x $. This theorem generalizes to the case of Markov chains with atoms.

---

### Theorem 37: Kac's Theorem for Chains with an Atom

Let $ (X_n) $ be $ \psi $-irreducible with an atom $ a $. The chain $ (X_n) $ is **positive recurrent** if and only if $ \mathbb{E}_a[T_a] < \infty $. In this case, the invariant distribution $ \pi $ for $ (X_n) $ satisfies:

$$
\pi(A) = \frac{\mathbb{E}_a[N_A]}{\mathbb{E}_a[T_a]},
$$

for any measurable set $ A $, where $ \mathbb{E}_a[N_A] $ denotes the expected number of visits to $ A $ before returning to $ a $.

---

### Proof Sketch

1. **Expected Return Time**:
   - Since $ a $ is an atom, $ \mathbb{E}_a[T_a] $ is well-defined and finite if the chain is positive recurrent.
   - The invariance of $ \pi $ implies that for any set $ A $, $ \pi(A) $ can be expressed as the ratio of expected visits to $ A $ over the expected return time to $ a $.

2. **Stationary Measure**:
   - The invariant measure satisfies:
     $$
     \pi(A) = \int_X K(x, A) \pi(dx),
     $$
     which is consistent with the definition involving $ \mathbb{E}_a[N_A] $.

3. **Uniqueness**:
   - Since $ (X_n) $ is $ \psi $-irreducible and has an atom, the invariant measure is unique (up to scaling).

4. **Positive Recurrence Condition**:
   - If $ \mathbb{E}_a[T_a] = \infty $, the chain is null recurrent or transient, and no finite invariant measure exists.

---

### Implications of Kac's Theorem

- **Eigenvalue Interpretation**:
  For discrete state-space Markov chains, $ \mathbb{E}_a[T_a]^{-1} $ is the eigenvector associated with eigenvalue 1 of the transition matrix $ P $.

- **MCMC Context**:
  Chains generated by Markov chain Monte Carlo methods are constructed to ensure positivity, guaranteeing the existence of a stationary distribution.

Kac's Theorem provides a direct link between the structure of the chain (via return times) and its stationary behavior, facilitating practical computations of invariant distributions in both theoretical and applied contexts.



```python
# Implementation of Kac's Theorem
# Computes the stationary distribution of a Markov chain using expected return times.

def compute_stationary_distribution(transition_matrix):
    """
    Computes the stationary distribution of a Markov chain using Kac's Theorem.

    Parameters:
        transition_matrix (list[list[float]]): A square matrix representing transition probabilities,
                                               where transition_matrix[i][j] is the probability of transitioning
                                               from state i to state j.

    Returns:
        list[float]: The stationary distribution as a list of probabilities.
    """
    num_states = len(transition_matrix)

    # Check if the matrix is square
    for row in transition_matrix:
        if len(row) != num_states:
            raise ValueError("Transition matrix must be square.")

    # Initialize expected return times for each state
    expected_return_times = [0.0] * num_states

    for state in range(num_states):
        visited = [False] * num_states
        queue = [(state, 0)]  # (current_state, steps)
        total_steps = 0
        visits = 0

        while queue:
            current_state, steps = queue.pop(0)
            if visited[current_state]:
                continue
            visited[current_state] = True
            visits += 1
            total_steps += steps

            for next_state in range(num_states):
                if not visited[next_state] and transition_matrix[current_state][next_state] > 0:
                    queue.append((next_state, steps + 1))

        if visits == 0:
            raise ValueError(f"State {state} is not recurrent.")

        expected_return_times[state] = total_steps / visits

    # Compute the stationary distribution using Kac's Theorem
    stationary_distribution = [1.0 / expected_return_times[state] for state in range(num_states)]

    # Normalize to ensure it sums to 1
    total = sum(stationary_distribution)
    stationary_distribution = [prob / total for prob in stationary_distribution]

    return stationary_distribution

# Example usage
if __name__ == "__main__":
    # Example transition matrix
    transition_matrix = [
        [0.5, 0.5, 0.0],
        [0.2, 0.5, 0.3],
        [0.0, 0.3, 0.7]
    ]

    stationary_dist = compute_stationary_distribution(transition_matrix)
    print("Stationary Distribution:", stationary_dist)
```

```python
import random
import matplotlib.pyplot as plt

def validate_transition_matrix(matrix):
    """Validate if the matrix is a proper transition matrix."""
    for row in matrix:
        if not (sum(row) - 1.0) < 1e-8:
            raise ValueError("Rows of the transition matrix must sum to 1.")
        if not all(0 <= prob <= 1 for prob in row):
            raise ValueError("Transition probabilities must be between 0 and 1.")

def simulate_return_time(matrix, state, iterations=10000):
    """Simulate the expected return time for a state."""
    n = len(matrix)
    total_time = 0
    successful_returns = 0

    for _ in range(iterations):
        current_state = state
        time_steps = 0

        while True:
            time_steps += 1
            next_state = random.choices(range(n), weights=matrix[current_state])[0]
            if next_state == state:
                total_time += time_steps
                successful_returns += 1
                break
            current_state = next_state

    if successful_returns == 0:
        return float('inf')
    return total_time / successful_returns

def kac_theorem_stationary_distribution(matrix, iterations=10000):
    """Compute the stationary distribution using Kac's theorem."""
    validate_transition_matrix(matrix)
    n = len(matrix)

    expected_return_times = [simulate_return_time(matrix, state, iterations) for state in range(n)]
    reciprocal_times = [1 / t if t != float('inf') else 0 for t in expected_return_times]

    total = sum(reciprocal_times)
    stationary_distribution = [value / total for value in reciprocal_times]

    return stationary_distribution, expected_return_times

def plot_stationary_distribution(distribution):
    """Plot the stationary distribution."""
    plt.figure(figsize=(8, 5))
    plt.bar(range(len(distribution)), distribution, color='skyblue', alpha=0.7)
    plt.xlabel('State')
    plt.ylabel('Probability')
    plt.title('Stationary Distribution (Kac\'s Theorem)')
    plt.show()

def plot_return_times(return_times):
    """Plot the expected return times."""
    plt.figure(figsize=(8, 5))
    plt.bar(range(len(return_times)), return_times, color='salmon', alpha=0.7)
    plt.xlabel('State')
    plt.ylabel('Expected Return Time')
    plt.title('Expected Return Times')
    plt.show()

# Example Transition Matrix
transition_matrix = [
    [0.5, 0.5, 0.0],
    [0.2, 0.5, 0.3],
    [0.0, 0.3, 0.7]
]

# Compute stationary distribution and return times
stationary_distribution, return_times = kac_theorem_stationary_distribution(transition_matrix)

# Print results
print("Stationary Distribution:", stationary_distribution)
print("Expected Return Times:", return_times)

# Plot results
plot_stationary_distribution(stationary_distribution)
plot_return_times(return_times)
```

### Positivity and Recurrence as Stability Properties

Positivity is a stability property stronger than recurrence. In fact, recurrence corresponds to the condition:

$$
P_a(T_a = \infty) = 0,
$$

which is a necessary condition for $E_a[a] < \infty$.

**Proof:**

If $E_a[a] < \infty$, then $P_a(T_a < \infty) = 1$. By Proposition 6.31, this implies that $(X)$ is recurrent.

Consider the measure given by:

$$
\sum_{n=1}^{\infty} P_a(X_n \in A, T_a < \infty).
$$

This measure is invariant since $P_a(T_a < \infty) = 1$. Now we compute:

$$
\int_0^\infty K(x, A) \, \mu(dx) = \int_0^\infty K(a, A) \, \mu(a) + \sum_{n=1}^{\infty} K(Z_n, A) P_a(T_a \geq 1).
$$

This results in:

$$
\int_0^\infty K(a, A) \, \mu(a) = \pi(A).
$$

It is finite since:

$$
\pi(\lambda) = \sum_{n=1}^{\infty} P_a(T_a \leq t).
$$

By uniqueness of the invariant distribution, we conclude that $E_a[T_a] < \infty$ implies that $\pi(A) = (E_a[T_a])^{-1}$.

Thus, the uniqueness of the invariant distribution follows, and renormalizing gives the desired result.

Assume $f'(1) \leq 1$. If there exists an invariant distribution for $(S_t)$, its characteristic function $g$ satisfies:

$$
g(\theta) = f(\theta) g(0) + g[f(\theta)] - g(0).
$$

In the simplest case, when the number of siblings of a given individual is distributed according to a Bernoulli distribution $B(p)$, we have:

$$
f(s) = q + ps, \quad \text{where} \quad q = 1 - p.
$$

The characteristic function $g(s)$ is a solution of:

$$
g(s) = q(q + ps) + p(s - 1) g(0).
$$

Iteratively substituting into the equation:

$$
g(s) = g(q + p(q + ps)) + p(q + ps - 1) g(0) + p(s - 1) g(U),
$$

for every $i \in \mathbb{N}$. Letting $k$ go to infinity:

$$
g(s) = g\left(\frac{q}{1 - p}\right) + \frac{p}{1 - p} (s - 1) g(0).

$$
g(s) = 1 + \frac{p}{1 - p} (s - 1) g(0).

Substituting $s = 0$ gives:

$$
g(0) = 1,
$$

and thus:

$$
g(s) = 1 + p(s - 1) = q + ps.
$$

Hence, the Bernoulli distribution is the invariant distribution, and the chain is positive.

In the case of a general Harris recurrent irreducible and aperiodic Markov chain $ (X) $ with stationary distribution $ \pi $, Hobert and Robert (2004) propose the following:

### Kac's Theorem and Its Representation

A representation of Kac's Theorem can be given in the form:

$$
(A) = \Pr(N_t \in A) \Pr(T^* = t)
$$

where $ T $ is an integer-valued random variable and, for each $ t \in \mathbb{N} $, $ N_t $ is a random variable whose distribution depends on $ t $. This representation is closely related to Kac's original theorem but offers a broader range of applicability, as it does not require the chain to have an atom. The only assumption is that the minorizing condition (6.10) is satisfied, and for simplicity, we take $ m = 1 $ in (6.10).

The random variables $ N $ and $ T $ can then be defined in terms of the split chain of Section 6.3.2. If $ T_C \geq 1 $ denotes the renewal time associated with the small set $ C $ in (6.10), then $ E[T_C] < \infty $ by recurrence (see Problem 6.32), and $ T $ is given by the tail probabilities of $ T_C $ as:

$$
P(T_t) = \Pr(T_C \geq t) \cdot E[T_C]
$$

The random variable $ N $ is then logically distributed from $ v $ if $ t = 1 $, and as $ X_t $, conditional on no renewal before time $ t $, otherwise, following from (6.10). Breyer and Roberts (2000b) derive this representation by means of a functional equation (see Problem 6.33).

Simulating from $ T $ thus amounts to simulating $ T^* $ according to (6.21) and then, for $ T = t $, simulating $ N $. Simulating the latter starts from the minorizing measure and then runs $ t - 1 $ steps of the residual distribution:

$$
K(z, \cdot) = K(z, \cdot) - e^{\ell C(x)} v(x)
$$

In cases when simulating from the residual is too complex, a brute-force Accept-Reject implementation is to run the split chain for $ t $ iterations until $ T_C > t $, though this may be too time-consuming in many situations. Hobert and Robert (2004) propose more advanced approaches to the simulation of $ T^* $.

### Reversibility and the Detailed Balance Condition

The stability property inherent to stationary chains can be related to another stability property called **reversibility**, which states that the direction of time...

### Reversibility and Detailed Balance Condition

The direction of time does not matter in the dynamics of the chain (see also Problems 6.53 and 6.54).

#### Definition 6.44: Reversibility
A stationary Markov chain $ (X_n) $ is **reversible** if the distribution of $ X_{n+1} $, conditionally on $ X_{n+2r} $, is the same as the distribution of $ X_{n+1} $ conditionally on $ X_1 = x $.

In fact, reversibility can be linked with the existence of a stationary measure if a condition stronger than in Definition 6.35 holds.

#### Definition 6.45: Detailed Balance Condition
A Markov chain with transition kernel $ K $ satisfies the **detailed balance condition** if there exists a function $ f $ satisfying:

$$
K(y, x) f(y) = K(x, y) f(x)
$$

for every $ (x, y) $.

While this condition is not necessary for $ f $ to be a stationary measure associated with the transition kernel $ K $, it provides a sufficient condition that is often easy to check and that can be used for most MCMC algorithms. The balance condition in equation (6.22) expresses an equilibrium in the flow of the Markov chain, namely that the probability of being in $ x $ and moving to $ y $ is the same as the probability of being in $ y $ and moving back to $ x $.

When $ f $ is a density, it also implies that the chain is **reversible**.

### Theorem 46: Detailed Balance and Reversibility

Suppose that a Markov chain with transition function $ K $ satisfies the detailed balance condition with a probability density function $ f $. Then:

1. **The density is the invariant density of the chain.**
2. **The chain is reversible.**

#### Proof:
Part (i) follows by noting that, by the detailed balance condition, for any measurable set $ R $:

$$
\int K(v, B) f(y) \, dy = \int K(y, x) f(x) \, dx
$$

Since $ \int f K(x, y) \, dy = 1 $, with the existence of the kernel $ K $ and the invariant density, it is clear that detailed balance and reversibility are the same property.

If $ f(x, y) $ is a joint density, then we can write (with obvious notation):

$$
f(x, y) = f_{xy}(x, y) / \gamma
$$

Thus, detailed balance requires that $ f_x = f_y $ and $ f_{xy} = f_{yx} $, meaning there is symmetry in the conditionals and the marginals are the same.

> **Note**: If there are no measure-theoretic difficulties with the definition of the kernel $ K $, both notions are equivalent.


```python
import numpy as np
import matplotlib.pyplot as plt

# Target distribution (e.g., Gaussian with mean 0, variance 1)
def target_distribution(x):
    return np.exp(-0.5 * x ** 2) / np.sqrt(2 * np.pi)

# Proposal distribution (Gaussian with mean 0, variance 1)
def proposal_distribution(x):
    return np.random.normal(x, 1)  # Propose new state from N(x, 1)

# Metropolis-Hastings algorithm implementation
def metropolis_hastings(num_samples, initial_state=0):
    samples = [initial_state]
    current_state = initial_state
    
    for _ in range(num_samples):
        proposed_state = proposal_distribution(current_state)
        
        # Calculate the acceptance ratio
        acceptance_ratio = min(1, target_distribution(proposed_state) / target_distribution(current_state))
        
        # Accept or reject the proposal
        if np.random.rand() < acceptance_ratio:
            current_state = proposed_state
        
        # Store the sample
        samples.append(current_state)
    
    return np.array(samples)

# Number of samples to generate
num_samples = 10000

# Generate samples using Metropolis-Hastings
samples = metropolis_hastings(num_samples)

# Plot the generated samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label='Samples')
x = np.linspace(-5, 5, 1000)
plt.plot(x, target_distribution(x), 'r-', label='Target Distribution (Normal(0, 1))')
plt.title('Metropolis-Hastings Sampling')
plt.legend()
plt.show()
```

```python
import random
import math
import sys

# Target distribution (Standard Normal Distribution: N(0,1))
def target_distribution(x):
    return math.exp(-0.5 * x ** 2) / math.sqrt(2 * math.pi)

# Proposal distribution (Gaussian with mean = current state, variance = 1)
def proposal_distribution(x):
    return random.gauss(x, 1)  # Propose new state from N(x, 1)

# Metropolis-Hastings algorithm implementation
def metropolis_hastings(num_samples, initial_state=0):
    samples = [initial_state]  # List to store the samples
    current_state = initial_state
    
    for _ in range(num_samples):
        proposed_state = proposal_distribution(current_state)
        
        # Calculate the acceptance ratio
        acceptance_ratio = min(1, target_distribution(proposed_state) / target_distribution(current_state))
        
        # Accept or reject the proposal based on the acceptance ratio
        if random.random() < acceptance_ratio:
            current_state = proposed_state
        
        samples.append(current_state)  # Store the sample
    
    return samples

# Plot function for text-based histogram (simple approach)
def plot_samples(samples, num_bins=50):
    min_sample = min(samples)
    max_sample = max(samples)
    bin_width = (max_sample - min_sample) / num_bins

    # Initialize bin counts
    bins = [0] * num_bins

    # Populate bins
    for sample in samples:
        bin_index = int((sample - min_sample) / bin_width)
        if bin_index >= num_bins:
            bin_index = num_bins - 1  # Ensure we don't go out of bounds
        bins[bin_index] += 1

    # Print histogram
    for i in range(num_bins):
        bin_range = f"{min_sample + i * bin_width:.2f} - {min_sample + (i + 1) * bin_width:.2f}"
        bar = "*" * bins[i]
        print(f"{bin_range}: {bar} ({bins[i]})")

# Number of samples to generate
num_samples = 1000

# Generate samples using the Metropolis-Hastings algorithm
samples = metropolis_hastings(num_samples)

# Plot the histogram (text-based)
plot_samples(samples)
```
