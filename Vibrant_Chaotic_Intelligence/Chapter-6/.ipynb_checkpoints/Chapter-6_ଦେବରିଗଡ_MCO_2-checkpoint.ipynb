## Proof and Explanation

The proof of the characterization result is quite involved (see Meyn and Tweedie 1993, pp. 107-109). The decomposition of $ X $ as a denumerable union of small sets is based on an arbitrary small set $ C $ and the sequence:

$$
C_m = \{ y; K^n(y, C) > 1/m \}
$$

Small sets are obviously easier to exhibit than atoms, given the freedom allowed by the minorizing condition (6.11). Moreover, they are, in fact, very common since, in addition to Theorem 6.21, Meyn and Tweedie (1993, p. 134) show that for sufficiently regular (in a topological sense) Markov chains, **every compact set is small**. Atoms, although a special case of small sets, enjoy stronger stability properties since the transition probability is invariant on $ a $. However, splitting methods (see below) offer the possibility of extending most of these properties to the general case and will be used as a technique of proof in the remainder of the chapter.

If the minorizing condition holds for $ X $, there are two ways of deriving a companion Markov chain $ X' $ sharing many properties with $ X $ and possessing an atom $ a $. The first method is called **Nummelin's splitting**, which constructs a chain made of two copies of $ X $ (see Nummelin 1978 and Meyn and Tweedie 1993, Section 5.1).

A second method, discovered around the same time, is due to **Athreya and Ney (1978)** and uses a stopping time to create an atom. We prefer to focus on this latter method because it is related to notions of **renewal time**, which are also useful in the control of Markov chain Monte Carlo algorithms (see Section 12.2.3).

### Definition 6.22: Renewal Time

A **renewal time** (or **regeneration time**) is a stopping rule $ \tau $ with the property that:

$$
(X_\tau, X_{\tau+1}, \ldots) \quad \text{is independent of} \quad (X_{0}, X_1, \ldots, X_{\tau-1})
$$

For instance, in Example 6.12, the returns to zero gain are renewal times. The excursions between two returns to zero are independent and identically distributed (see Feller 1970, Chapter III). More generally, visits to atoms are renewal times, whose features are quite appealing in convergence control for Markov chain Monte Carlo algorithms (see Chapter 12).

If condition (6.10) holds, and if the probability $ \Pr(T_C < \infty) $ of a return to $ C $ in a finite time is identically equal to 1 on $ X $, **Athreya and Ney (1978)** modify the transition kernel when $ X \in C $ by simulating $ X_{n+1} $ as:

$$
X_{n+1} = 
\begin{cases} 
X_{n} & \text{with probability } 1 - \epsilon \\
\text{Simulate } X_{n+1} & \text{with probability } \epsilon
\end{cases}
$$

That is, by simulating $ X_{n+1} $ from $ K(X) $ with probability $ \epsilon $ every time $ X_n $ is in $ C $. This modification does not change the marginal distribution of $ X_{n+1} $ conditionally on $ X_n $, since:

$$
\Pr(X_{n+1} = x | X_n = x) = K(x, x) \quad \text{and} \quad \Pr(X_{n+1} = y | X_n = x) = K(x, y) \quad \text{for} \quad y \neq x.
$$

This method of modifying the transition kernel can help create atoms and regulate convergence in Markov Chain Monte Carlo algorithms.
## Renewal Times and Split Chains

The following equation results from modifying the transition kernel:

$$
\epsilon v(A) + (1 - \epsilon) K(X_n, A) - \epsilon v(A) = K(x, A), \quad \forall A \in \mathcal{B}(X),
$$

where $ \epsilon $ is the probability that $ X_{n+1} $ is simulated from a measure $ \nu $, and it produces **renewal times** for each time $ j $ such that $ X_j \in C $ and $ X_{j+1} \sim \nu $. 

We can clearly see how the renewal times result in independent chains. When $ X_{j+1} \sim \nu $, this event is totally independent of any past history, as the current state of the chain has no effect on the measure $ \nu $. Note also the key role played by the **minorization condition**. It allows us to create the split chain with the same marginal distribution as the original chain.

We define the sequence of **renewal times** $ \{ T_n \} $ as:

$$
T_n = \inf \{ n > T_{n-1}; X_n \in C \text{ and } X_{n+1} \sim \nu \}
$$

with $ T_0 = 0 $. Athreya and Ney (1978) introduce the **augmented chain**, also called the **split chain**:

$$
X_n = (X_n, n), \quad \text{with} \quad n = 1 \text{ when } X_n \in C \text{ and } X_{n+1} \sim \nu.
$$

It is then easy to show that the set $ C \times \{1\} $ is an **atom** of the chain $ (X) $, and the resulting subchain $ (X) $ is still a Markov chain with transition kernel $ K(n, \cdot) $.

The notion of small sets is useful primarily in finite and discrete settings where the individual probabilities of states are too small to allow for a reasonable rate of renewal. In these cases, small sets are made of collections of states with a minimum probability. Otherwise, small sets that reduce to a single value are also considered atoms.

## Cycles and Aperiodicity

The behavior of $ (X) $ may sometimes be restricted by deterministic constraints on the moves from $ X_n $ to $ X_{n+1} $. We formalize these constraints here and show in the following chapters that the chains produced by Markov chain Monte Carlo algorithms do not display this behavior and, hence, do not suffer from the associated drawbacks.

In the discrete case, the **period** of a state $ w \in X $ is defined as:

$$
d(w) = \text{gcd} \left\{ m \geq 1; P^m(w, w) > 0 \right\}
$$

where gcd denotes the greatest common divisor. The value of the period is constant for all states that communicate with $ w $. In the case of an irreducible chain on a finite space $ X $, the transition matrix can be written (with a possible reordering of the states) as a block matrix:

$$
P = \begin{pmatrix}
0 & D & 0 & 0 & 0 & D_2 & 0 \\
0 & 0 & 0 & D_1 & 0 & 0 & 0 \\
\end{pmatrix}
$$

## Cycles and Aperiodicity (Continued)

The blocks $ D_i $ in the transition matrix are stochastic matrices. This representation clearly illustrates the forced passage from one group of states to another, with a return to the initial group occurring every $ d $-th step. If the chain is **irreducible** (so all states communicate), there is only one value for the period. An irreducible chain is **aperiodic** if it has period 1. The extension to the general case requires the existence of a **small set**.

### Definition .23: A-irreducible Chain

An $ A $-irreducible chain $ (X) $ has a cycle of length $ d $ if there exists a small set $ C $, an associated integer $ M $, and a probability distribution $ v_M $ such that $ d $ is the greatest common divisor (g.c.d.) of

$$
d = \gcd \left\{ m \geq 1; P^m(C, C) > 0 \text{ for some } m \geq 8m \cdot v_M \right\}.
$$

A decomposition like equation (6.13) can be established in general. It is easily shown that the number $ d $ is independent of the small set $ C $ and that this number intrinsically characterizes the chain $ (X) $. The period of $ (X) $ is then defined as the largest integer $ d $ satisfying Definition 6.23, and $ (X_n) $ is aperiodic if $ d = 1 $.

If there exists a small set $ A $ and a minorizing measure such that $ v_1(A) > 0 $ (so it is possible to go from $ A $ to $ A $ in a single step), the chain is said to be **strongly aperiodic**. Note that the $ K $-chain can be used to transform an aperiodic chain into a strongly aperiodic chain.

### Example .24: Bernoulli-Laplace Chain

**(Continuation of Example .14)**

The Bernoulli-Laplace chain is aperiodic and even strongly aperiodic since the diagonal terms satisfy $ P_{ii} > 0 $ for every $ i = \{ 0, \dots, K \} $.

## Transience and Recurrence

###  Classification of Irreducible Chains

From an algorithmic point of view, a **Markov chain** must enjoy good stability properties to guarantee an acceptable approximation of the simulated model. Indeed, **irreducibility** ensures that every set $ A $ will be visited by the Markov chain.


```python
import numpy as np

class MarkovChain:
    def __init__(self, transition_matrix):
        """
        Initialize the Markov chain with a given transition matrix.
        """
        self.transition_matrix = np.array(transition_matrix)
        self.num_states = self.transition_matrix.shape[0]

    def is_aperiodic(self):
        """
        Check if the Markov chain is aperiodic.
        The chain is aperiodic if it has period 1.
        """
        def gcd(a, b):
            while b:
                a, b = b, a % b
            return a

        period = np.inf
        for m in range(1, 100):  # Try up to 100 steps
            reachable = np.sum(np.linalg.matrix_power(self.transition_matrix, m), axis=1) > 0
            if all(reachable):
                period = gcd(period, m)
                if period == 1:
                    return True
        return period == 1

    def is_irreducible(self):
        """
        Check if the Markov chain is irreducible.
        The chain is irreducible if all states communicate.
        """
        reachable = np.sum(np.linalg.matrix_power(self.transition_matrix, self.num_states), axis=1) > 0
        return all(reachable)

    def simulate(self, initial_state, steps):
        """
        Simulate the Markov chain for a given number of steps starting from an initial state.
        """
        state = initial_state
        states_visited = [state]
        for _ in range(steps):
            state = np.random.choice(self.num_states, p=self.transition_matrix[state])
            states_visited.append(state)
        return states_visited

# Example: A simple 3-state Markov Chain
transition_matrix = [
    [0.5, 0.5, 0.0],
    [0.2, 0.5, 0.3],
    [0.0, 0.3, 0.7]
]

# Create Markov Chain instance
mc = MarkovChain(transition_matrix)

# Check irreducibility and aperiodicity
print("Is the chain irreducible?", mc.is_irreducible())
print("Is the chain aperiodic?", mc.is_aperiodic())

# Simulate the chain for 10 steps starting from state 0
initial_state = 0
steps = 10
states_visited = mc.simulate(initial_state, steps)
print(f"States visited in {steps} steps: {states_visited}")
```

```python
import random

class MarkovChain:
    def __init__(self, transition_matrix):
        """
        Initialize the Markov chain with a given transition matrix.
        """
        self.transition_matrix = transition_matrix
        self.num_states = len(transition_matrix)

    def is_aperiodic(self):
        """
        Check if the Markov chain is aperiodic.
        The chain is aperiodic if it has period 1.
        """
        def gcd(a, b):
            while b:
                a, b = b, a % b
            return a

        period = float('inf')
        for m in range(1, 100):  # Try up to 100 steps
            reachable = [False] * self.num_states
            for i in range(self.num_states):
                if self._reachable_from(i, m):
                    reachable[i] = True
            if all(reachable):
                period = gcd(period, m)
                if period == 1:
                    return True
        return period == 1

    def _reachable_from(self, state, m):
        """
        Check if there is a return to the same state in m steps.
        """
        current_state = state
        for _ in range(m):
            next_state = self._next_state(current_state)
            current_state = next_state
        return current_state == state

    def is_irreducible(self):
        """
        Check if the Markov chain is irreducible.
        The chain is irreducible if all states communicate.
        """
        for state in range(self.num_states):
            reachable = [False] * self.num_states
            for m in range(1, self.num_states + 1):
                for next_state in range(self.num_states):
                    if self._reachable_from(state, m):
                        reachable[next_state] = True
            if not all(reachable):
                return False
        return True

    def simulate(self, initial_state, steps):
        """
        Simulate the Markov chain for a given number of steps starting from an initial state.
        """
        state = initial_state
        states_visited = [state]
        for _ in range(steps):
            state = self._next_state(state)
            states_visited.append(state)
        return states_visited

    def _next_state(self, current_state):
        """
        Given the current state, return the next state based on the transition matrix.
        """
        prob = random.random()
        cumulative_prob = 0.0
        for next_state, p in enumerate(self.transition_matrix[current_state]):
            cumulative_prob += p
            if cumulative_prob > prob:
                return next_state
        return len(self.transition_matrix) - 1  # fallback in case of rounding errors

# Example: A simple 3-state Markov Chain
transition_matrix = [
    [0.5, 0.5, 0.0],
    [0.2, 0.5, 0.3],
    [0.0, 0.3, 0.7]
]

# Create Markov Chain instance
mc = MarkovChain(transition_matrix)

# Check irreducibility and aperiodicity
print("Is the chain irreducible?", mc.is_irreducible())
print("Is the chain aperiodic?", mc.is_aperiodic())

# Simulate the chain for 10 steps starting from state 0
initial_state = 0
steps = 10
states_visited = mc.simulate(initial_state, steps)
print(f"States visited in {steps} steps: {states_visited}")
```

```python
import random
import matplotlib.pyplot as plt

class MarkovChain:
    def __init__(self, transition_matrix):
        """
        Initialize the Markov chain with a given transition matrix.
        """
        self.transition_matrix = transition_matrix
        self.num_states = len(transition_matrix)

    def is_aperiodic(self):
        """
        Check if the Markov chain is aperiodic.
        The chain is aperiodic if it has period 1.
        """
        def gcd(a, b):
            while b:
                a, b = b, a % b
            return a

        period = float('inf')
        for m in range(1, 100):  # Try up to 100 steps
            reachable = [False] * self.num_states
            for i in range(self.num_states):
                if self._reachable_from(i, m):
                    reachable[i] = True
            if all(reachable):
                period = gcd(period, m)
                if period == 1:
                    return True
        return period == 1

    def _reachable_from(self, state, m):
        """
        Check if there is a return to the same state in m steps.
        """
        current_state = state
        for _ in range(m):
            next_state = self._next_state(current_state)
            current_state = next_state
        return current_state == state

    def is_irreducible(self):
        """
        Check if the Markov chain is irreducible.
        The chain is irreducible if all states communicate.
        """
        for state in range(self.num_states):
            reachable = [False] * self.num_states
            for m in range(1, self.num_states + 1):
                for next_state in range(self.num_states):
                    if self._reachable_from(state, m):
                        reachable[next_state] = True
            if not all(reachable):
                return False
        return True

    def simulate(self, initial_state, steps):
        """
        Simulate the Markov chain for a given number of steps starting from an initial state.
        """
        state = initial_state
        states_visited = [state]
        for _ in range(steps):
            state = self._next_state(state)
            states_visited.append(state)
        return states_visited

    def _next_state(self, current_state):
        """
        Given the current state, return the next state based on the transition matrix.
        """
        prob = random.random()
        cumulative_prob = 0.0
        for next_state, p in enumerate(self.transition_matrix[current_state]):
            cumulative_prob += p
            if cumulative_prob > prob:
                return next_state
        return len(self.transition_matrix) - 1  # fallback in case of rounding errors

    def plot_state_visits(self, states_visited):
        """
        Plot the visited states over time.
        """
        plt.figure(figsize=(10, 6))
        plt.plot(states_visited, marker='o', linestyle='-', color='b')
        plt.title('States Visited Over Time')
        plt.xlabel('Time Step')
        plt.ylabel('State')
        plt.grid(True)
        plt.show()

    def plot_state_frequencies(self, states_visited):
        """
        Plot the frequency of visits to each state.
        """
        state_counts = [states_visited.count(i) for i in range(self.num_states)]
        
        plt.figure(figsize=(10, 6))
        plt.bar(range(self.num_states), state_counts, color='orange')
        plt.title('State Visit Frequencies')
        plt.xlabel('State')
        plt.ylabel('Frequency')
        plt.xticks(range(self.num_states))
        plt.grid(True)
        plt.show()

    def plot_transition_matrix(self):
        """
        Plot the transition matrix as a heatmap.
        """
        import seaborn as sns
        
        plt.figure(figsize=(8, 6))
        sns.heatmap(self.transition_matrix, annot=True, cmap='Blues', cbar=True)
        plt.title('Transition Matrix')
        plt.xlabel('To State')
        plt.ylabel('From State')
        plt.show()

# Example: A simple 3-state Markov Chain
transition_matrix = [
    [0.5, 0.5, 0.0],
    [0.2, 0.5, 0.3],
    [0.0, 0.3, 0.7]
]

# Create Markov Chain instance
mc = MarkovChain(transition_matrix)

# Simulate the chain for 100 steps starting from state 0
initial_state = 0
steps = 100
states_visited = mc.simulate(initial_state, steps)

# Plot graphs
mc.plot_state_visits(states_visited)  # Plot states visited over time
mc.plot_state_frequencies(states_visited)  # Plot state visit frequencies
mc.plot_transition_matrix()  # Plot transition matrix as heatmap
```
