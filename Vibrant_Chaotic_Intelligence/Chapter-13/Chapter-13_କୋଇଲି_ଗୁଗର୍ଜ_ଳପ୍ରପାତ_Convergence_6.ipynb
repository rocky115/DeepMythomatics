# Perfect Sampling via Automatic Coupling

## Introduction to Perfect Sampling

In addition to slice sampling, there is another approach to perfect sampling, which is interesting from a theoretical point of view, as it highlights the connection with Accept-Reject algorithms. It is based on Boyer-Moore diagrams (2000) (automata coupling).

We have two chains $(c_t^{(1)})$ and $(c_t^{(2)})$, where the first chain, $(c_t^{(1)})$, is run following a standard Metropolis-Hastings scheme with proposal density $q(x'|x)$ and corresponding acceptance probability:






$$\alpha_1(x'|x) = \min\left(1, \frac{\pi(x')q(x|x')}{\pi(x)q(x'|x)}\right)$$

The transition for the second chain, $(c_t^{(2)})$, is based on the same proposed value $x'$, with the difference being that the proposed distribution is independent of the second chain and is associated with the acceptance probability:

$$\alpha_2(x'|x) = \min\left(1, \frac{\pi(x')}{\pi(x)}\right)$$

If $\alpha_1(x'|x) = \alpha_2(x'|x)$, then we would have the same behavior for both chains. Then, keeping in mind the perfect sampling perspective, we can use the coupling time to claim that the two chains start to coalesce once they both accept a proposed value $x'$.

## Mathematical Foundation

The coupling of chains provides a theoretical guarantee for perfect sampling from the target distribution. By observing when two chains with different starting points begin to follow identical paths (coalescence), we can ensure samples are drawn from the stationary distribution.

## Implementation Considerations

When implementing this approach:

1. Initialize two chains at different starting points
2. Use the same proposed values for transitions in both chains
3. Apply the respective acceptance probabilities ($\alpha_1$ and $\alpha_2$)
4. Monitor when the chains coalesce
5. After coalescence, samples can be considered as drawn from the target distribution

## Advantages and Limitations

**Advantages:**
- Theoretically sound approach to perfect sampling
- Highlights connection with Accept-Reject algorithms
- Provides guaranteed samples from the target distribution

**Limitations:**
- May require long run times before coalescence occurs
- Implementation can be more complex than standard MCMC methods
- Computational efficiency may be lower than other sampling approaches



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

class PerfectSamplingAutomaticCoupling:
    """
    Implementation of Perfect Sampling via Automatic Coupling.
    
    This class implements the perfect sampling algorithm using automatic coupling
    as described in Boyer-Moore diagrams (2000) and Casella et al. (2002).
    """
    
    def __init__(self, target_density, proposal_sampler, proposal_density=None, 
                 max_iterations=10000, seed=None):
        """
        Initialize the perfect sampler.
        
        Parameters:
        -----------
        target_density : function
            Unnormalized target density function π(x)
        proposal_sampler : function
            Function to sample from proposal distribution q(·|x)
        proposal_density : function, optional
            Function to compute proposal density q(x'|x). If None, assumes symmetric proposal.
        max_iterations : int, optional
            Maximum number of iterations to try before giving up
        seed : int, optional
            Random seed for reproducibility
        """
        self.target_density = target_density
        self.proposal_sampler = proposal_sampler
        self.proposal_density = proposal_density
        self.max_iterations = max_iterations
        
        # Set random seed if provided
        if seed is not None:
            np.random.seed(seed)
            
    def _compute_alpha1(self, current_x, proposed_x):
        """
        Compute the acceptance probability for the first chain (standard MH).
        
        α₁(x'|x) = min(1, (π(x')q(x|x'))/(π(x)q(x'|x)))
        """
        target_ratio = self.target_density(proposed_x) / self.target_density(current_x)
        
        # Handle symmetric proposal case
        if self.proposal_density is None:
            return min(1.0, target_ratio)
        
        # Compute proposal ratio for asymmetric proposal
        proposal_ratio = self.proposal_density(current_x, proposed_x) / self.proposal_density(proposed_x, current_x)
        return min(1.0, target_ratio * proposal_ratio)
    
    def _compute_alpha2(self, current_x, proposed_x):
        """
        Compute the acceptance probability for the second chain.
        
        α₂(x'|x) = min(1, π(x')/π(x))
        """
        target_ratio = self.target_density(proposed_x) / self.target_density(current_x)
        return min(1.0, target_ratio)
    
    def sample(self, initial_x1, initial_x2):
        """
        Generate a sample using perfect sampling via automatic coupling.
        
        Parameters:
        -----------
        initial_x1 : float or array
            Initial value for chain 1
        initial_x2 : float or array
            Initial value for chain 2
            
        Returns:
        --------
        sample : float or array
            Perfect sample from target distribution
        coalescence_time : int
            Number of iterations until coalescence
        chain1_states : list
            States of chain 1
        chain2_states : list
            States of chain 2
        """
        # Initialize the chains
        x1 = initial_x1
        x2 = initial_x2
        
        # Keep track of chains' states for plotting
        chain1_states = [x1]
        chain2_states = [x2]
        
        # Set up coalescence detection
        has_coalesced = False
        t = 0
        
        while not has_coalesced and t < self.max_iterations:
            # Generate a proposal (same for both chains)
            proposed_x = self.proposal_sampler(x1)
            
            # Compute acceptance probabilities
            alpha1 = self._compute_alpha1(x1, proposed_x)
            alpha2 = self._compute_alpha2(x2, proposed_x)
            
            # Accept/reject for chain 1
            if np.random.rand() < alpha1:
                x1 = proposed_x
            
            # Accept/reject for chain 2
            if np.random.rand() < alpha2:
                x2 = proposed_x
                
            # Check for coalescence
            has_coalesced = np.array_equal(x1, x2)
            
            # Store chain states
            chain1_states.append(x1)
            chain2_states.append(x2)
            
            t += 1
            
        if t == self.max_iterations and not has_coalesced:
            print(f"Warning: Chains did not coalesce within {self.max_iterations} iterations.")
        else:
            print(f"Chains coalesced after {t} iterations.")
            
        return x1, t, chain1_states, chain2_states

    def visualize_chains(self, chain1_states, chain2_states, coalescence_time):
        """
        Visualize the evolution of the two chains.
        
        Parameters:
        -----------
        chain1_states : list
            States of chain 1
        chain2_states : list
            States of chain 2
        coalescence_time : int
            Time to coalescence
        """
        plt.figure(figsize=(10, 6))
        
        # Convert to numpy arrays for easier handling
        chain1 = np.array(chain1_states)
        chain2 = np.array(chain2_states)
        
        # Plot the chains
        plt.plot(chain1, 'b-', alpha=0.7, label='Chain 1')
        plt.plot(chain2, 'r-', alpha=0.7, label='Chain 2')
        
        # Highlight coalescence
        if coalescence_time < len(chain1):
            plt.axvline(x=coalescence_time, color='g', linestyle='--', 
                     label=f'Coalescence at t={coalescence_time}')
        
        plt.title('Perfect Sampling via Automatic Coupling')
        plt.xlabel('Iteration')
        plt.ylabel('Parameter value')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()


# Example usage:
if __name__ == "__main__":
    # Example: Sampling from a mixture of Gaussians
    def target_density(x):
        """Unnormalized density of a mixture of Gaussians"""
        return 0.4 * stats.norm.pdf(x, -2, 0.8) + 0.6 * stats.norm.pdf(x, 2, 1.2)
    
    def proposal_sampler(x):
        """Random walk proposal"""
        return x + np.random.normal(0, 1.5)
    
    def proposal_density(x_proposed, x_current):
        """Density of random walk proposal"""
        return stats.norm.pdf(x_proposed - x_current, 0, 1.5)
    
    # Initialize sampler
    perfect_sampler = PerfectSamplingAutomaticCoupling(
        target_density=target_density,
        proposal_sampler=proposal_sampler,
        proposal_density=proposal_density,
        max_iterations=5000,
        seed=42
    )
    
    # Generate a perfect sample
    initial_x1 = -5.0  # Different starting points
    initial_x2 = 5.0
    
    sample, coalescence_time, chain1_states, chain2_states = perfect_sampler.sample(
        initial_x1, initial_x2
    )
    
    # Visualize the chains
    perfect_sampler.visualize_chains(chain1_states, chain2_states, coalescence_time)
    
    # Generate multiple samples to check distribution
    n_samples = 1000
    samples = []
    
    for i in range(n_samples):
        if i % 100 == 0:
            print(f"Generating sample {i}/{n_samples}")
            
        # Use different starting points for each run
        initial_x1 = -5.0 + np.random.normal(0, 1)
        initial_x2 = 5.0 + np.random.normal(0, 1)
        
        sample, _, _, _ = perfect_sampler.sample(initial_x1, initial_x2)
        samples.append(sample)
    
    # Plot the resulting distribution
    plt.figure(figsize=(10, 6))
    
    # Create grid for true density
    x_grid = np.linspace(-6, 6, 1000)
    true_density = np.array([target_density(x) for x in x_grid])
    true_density = true_density / np.trapz(true_density, x_grid)  # Normalize
    
    # Plot histogram of samples
    sns.histplot(samples, stat='density', alpha=0.6, label='Perfect samples')
    
    # Plot true density
    plt.plot(x_grid, true_density, 'r-', label='True density')
    
    plt.title('Distribution of Perfect Samples')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

# The Rudimentary Coupling Strategy for Perfect Sampling

## Introduction to Coupling Strategy

The implications of this rudimentary coupling strategy on perfect sampling are far-reaching. Consider the special case of a 1-dimensional proposal density $b$ for the first chain $(c_t^{(1)})$. Starting from $c_t^{(1)}$ with candidate $y$, the acceptance probabilities simplify to:

$$\alpha_1^{(y)} = \frac{\pi(y)b(c_t^{(1)}|y)}{\pi(c_t^{(1)})b(y|c_t^{(1)})} \wedge 1 \quad \text{and} \quad \alpha_2^{(y)} = \frac{\pi(y)b(c_t^{(1)}|y)}{\pi(c_t^{(1)})b(y|c_t^{(1)})} \wedge 1$$

We can then introduce the (total) ordering associated with $\pi/b$, that is:

$$x \ll z \Leftrightarrow \frac{\pi(x)}{b(x)} \le \frac{\pi(z)}{b(z)}$$

and observe that the above coupling strategy preserves the ordering (Problem 13.7). This property, namely that the coupling proposal preserves the ordering, was noted by Corcoran and Tweedie (2002).

If one is able to identify in advance the values of the states that maximize $\pi/b$ and $0$ and $1$, not the order $\ll$, we can then run a perfect sampler based on just two chains starting from $0$ and $1$. Note that, since the choice of $0$ is arbitrary, a sampler can be oriented towards an easy derivation of $0$ and $1$. Nonetheless, this cannot be achieved in every setting, since the existence of $1$ implies that $\pi/b$ has a maximum. This is not necessarily the case when $b$ has unbounded support, as shown in Theorem 7.8. In the special case where the state space $\mathcal{X}$ is compact, $1$ can be chosen as the uniform distribution over the extremal elements $0$ and $1$ are then those induced by $\pi/b$.

## Example 13.13 (Continuation of Example 13.12)

The strategy applies when both chains $(c_t^{(1)})$ and $(c_t^{(2)})$ under the pair $(m, n)$ evolve in a coupled (finite) state. As an illustration, set $m$ to the value of $c_0$ and similarly define $n$ from the density proportional to:

$$(13.8) \quad \frac{f(x_i; 1, n_0)}{(1/k + n_0)^{n_0+1}} \cdot \frac{f(x_i; 1, n - n_0)}{(1/k + n-n_0)^{n-n_0+1}}$$

on the finite set of the possible values of:

$$x_0 = \sum_{j=1}^k \pi_j$$

The second chain $(c_t^{(2)})$ is the Markov chain with transition density defined in time according to a regular Metropolis-Hastings scheme with univariate proposal on the set of possible values of $x_0$. The chain $(c_t^{(2)})$, starting from $1$, is then coupled to the first chain by the above coupling scheme. If coalescence had not occurred by time $t$, we increase $t$.

If there is a maximum, coupling is perfect, while if the support is unbounded (not $[0,1]$), coupling is rejected by $(c_t^{(1)})$, which remains equal to the starting value $c_0^{(1)}$ unless a good jumping ratio $\frac{\pi(y)}{b(y|c_t^{(1)})}$ occurs, which has occurred, both chains coalesce.

## Mathematical Formulation of Coupling

The coupling strategy provides us with a mechanism to ensure perfect sampling by running two chains simultaneously and observing when they coalesce. Let's formalize the approach:

1. Run two chains $(c_t^{(1)})$ and $(c_t^{(2)})$ with different starting points
2. Use the same proposal distribution $b$ for both chains
3. Apply the acceptance probabilities $\alpha_1^{(y)}$ and $\alpha_2^{(y)}$ respectively
4. Monitor for coalescence of the chains

The key property is that the ordering is preserved through this coupling process, which allows us to make statements about the convergence behavior.

## Implications for Implementation

When implementing this coupling strategy:

1. Identify the extremal elements induced by $\pi/b$ if possible
2. Initialize chains at these extremal points
3. Apply the coupled transition mechanism
4. Sample from the chain after coalescence

This approach guarantees that samples are drawn from the target distribution $\pi$.

## Theoretical Guarantees

The ordering property ($\ll$) provides theoretical guarantees about the behavior of the coupled chains:

1. If $x \ll z$, then the chain starting from $x$ will reach any state at least as quickly as the chain starting from $z$
2. Coalescence of chains starting from extremal points guarantees coalescence for all possible starting points
3. After coalescence, the resulting state is distributed according to the target distribution $\pi$

## Practical Considerations

In practice, several issues need to be addressed:

1. Identifying extremal elements may be difficult for complex distributions
2. Unbounded support can lead to non-existence of maximum elements
3. Computational efficiency depends on coalescence time
4. For high-dimensional problems, coalescence might take prohibitively long

## Conclusion

The rudimentary coupling strategy presents an elegant theoretical approach to perfect sampling. When applicable, it provides guaranteed samples from the target distribution without requiring burn-in periods typical of standard MCMC methods.

The ordering property is central to the approach, allowing us to reduce the monitoring of an infinite number of chains to just two chains starting from extremal points. However, practical implementation may face challenges when extremal elements cannot be easily identified or when dealing with unbounded support.



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from typing import Callable, Tuple, List, Optional

class RudimentaryCouplingPerfectSampler:
    """
    Implementation of the rudimentary coupling strategy for perfect sampling.
    
    This class implements the perfect sampling algorithm based on the rudimentary
    coupling strategy using the ordering associated with π/b as described in
    Corcoran and Tweedie (2002).
    """
    
    def __init__(self, 
                 target_density: Callable,
                 proposal_density: Callable,
                 proposal_sampler: Callable,
                 min_state: float,
                 max_state: float,
                 max_iterations: int = 10000,
                 seed: Optional[int] = None):
        """
        Initialize the perfect sampler with rudimentary coupling.
        
        Parameters:
        -----------
        target_density : function
            Unnormalized target density function π(x)
        proposal_density : function
            Function to compute proposal density b(y|x)
        proposal_sampler : function
            Function to sample from proposal distribution b(·|x)
        min_state : float
            Minimum state value (corresponds to '0' in the ordering)
        max_state : float
            Maximum state value (corresponds to '1' in the ordering)
        max_iterations : int, optional
            Maximum number of iterations before giving up
        seed : int, optional
            Random seed for reproducibility
        """
        self.target_density = target_density
        self.proposal_density = proposal_density
        self.proposal_sampler = proposal_sampler
        self.min_state = min_state
        self.max_state = max_state
        self.max_iterations = max_iterations
        
        # Set random seed if provided
        if seed is not None:
            np.random.seed(seed)
    
    def _compute_alpha(self, current_x: float, proposed_y: float) -> float:
        """
        Compute the acceptance probability:
        
        α(y) = [π(y)b(x|y)]/[π(x)b(y|x)] ∧ 1
        
        Parameters:
        -----------
        current_x : float
            Current state
        proposed_y : float
            Proposed state
            
        Returns:
        --------
        alpha : float
            Acceptance probability
        """
        target_ratio = self.target_density(proposed_y) / self.target_density(current_x)
        proposal_ratio = self.proposal_density(current_x, proposed_y) / self.proposal_density(proposed_y, current_x)
        return min(1.0, target_ratio * proposal_ratio)
    
    def _is_less_in_ordering(self, x: float, z: float) -> bool:
        """
        Check if x ≪ z in the ordering associated with π/b.
        
        Parameters:
        -----------
        x : float
            First state
        z : float
            Second state
            
        Returns:
        --------
        bool : 
            True if x ≪ z, False otherwise
        """
        # Compute π(x)/b(x) and π(z)/b(z)
        # Here we use a base point for the proposal density
        base_point = (x + z) / 2  # Arbitrary base point for computing ratios
        ratio_x = self.target_density(x) / self.proposal_density(x, base_point)
        ratio_z = self.target_density(z) / self.proposal_density(z, base_point)
        
        return ratio_x <= ratio_z
    
    def sample(self) -> Tuple[float, int, List[float], List[float]]:
        """
        Generate a sample using the rudimentary coupling strategy.
        
        Returns:
        --------
        sample : float
            Perfect sample from target distribution
        coalescence_time : int
            Number of iterations until coalescence
        chain1_states : list
            States of chain 1 (starting from min_state)
        chain2_states : list
            States of chain 2 (starting from max_state)
        """
        # Initialize the chains at the extremal points
        x1 = self.min_state  # Chain 1 starts at '0'
        x2 = self.max_state  # Chain 2 starts at '1'
        
        # Keep track of chains' states for plotting and analysis
        chain1_states = [x1]
        chain2_states = [x2]
        
        # Set up coalescence detection
        has_coalesced = False
        t = 0
        
        while not has_coalesced and t < self.max_iterations:
            # Generate proposals (same for both chains)
            # We use x1 as the conditioning state for both chains to ensure coupling
            proposed_y = self.proposal_sampler(x1)
            
            # Compute acceptance probability (same formula for both chains)
            alpha1 = self._compute_alpha(x1, proposed_y)
            alpha2 = self._compute_alpha(x2, proposed_y)
            
            # Generate uniform random variables for acceptance decisions
            u1 = np.random.rand()
            u2 = np.random.rand()
            
            # Update chain 1
            if u1 < alpha1:
                x1 = proposed_y
            
            # Update chain 2
            if u2 < alpha2:
                x2 = proposed_y
            
            # Check for coalescence
            has_coalesced = np.isclose(x1, x2)
            
            # Store chain states
            chain1_states.append(x1)
            chain2_states.append(x2)
            
            t += 1
        
        if t == self.max_iterations and not has_coalesced:
            print(f"Warning: Chains did not coalesce within {self.max_iterations} iterations.")
        else:
            print(f"Chains coalesced after {t} iterations.")
        
        # After coalescence, both chains have the same state
        # which is a perfect sample from the target distribution
        return x1, t, chain1_states, chain2_states
    
    def verify_ordering_preservation(self, num_points: int = 100) -> None:
        """
        Verify that the coupling strategy preserves the ordering.
        
        Parameters:
        -----------
        num_points : int, optional
            Number of random points to check
            
        Returns:
        --------
        None
            Prints verification results
        """
        # Generate random points in the state space
        points = np.linspace(self.min_state, self.max_state, num_points)
        
        # Check ordering preservation for random pairs
        num_pairs = 50
        preserved_count = 0
        
        for _ in range(num_pairs):
            # Pick two random points
            idx1, idx2 = np.random.choice(num_points, 2, replace=False)
            x = points[idx1]
            z = points[idx2]
            
            # Check if x ≪ z in the original ordering
            original_ordering = self._is_less_in_ordering(x, z)
            
            # Generate proposal and compute acceptance probabilities
            proposed_y = self.proposal_sampler(x)
            alpha_x = self._compute_alpha(x, proposed_y)
            alpha_z = self._compute_alpha(z, proposed_y)
            
            # Check if the ordering is preserved after potential transitions
            if original_ordering:
                # If x ≪ z then either both accept or both reject
                preserved = (alpha_x <= alpha_z)
            else:
                # If z ≪ x then either both accept or both reject
                preserved = (alpha_z <= alpha_x)
            
            if preserved:
                preserved_count += 1
        
        print(f"Ordering preservation verified for {preserved_count}/{num_pairs} pairs.")
        print(f"Preservation rate: {preserved_count/num_pairs:.2f}")
    
    def visualize_chains(self, chain1_states: List[float], chain2_states: List[float], 
                       coalescence_time: int) -> None:
        """
        Visualize the evolution of the two chains.
        
        Parameters:
        -----------
        chain1_states : list
            States of chain 1
        chain2_states : list
            States of chain 2
        coalescence_time : int
            Time to coalescence
        """
        plt.figure(figsize=(12, 6))
        
        # Convert to numpy arrays for easier handling
        chain1 = np.array(chain1_states)
        chain2 = np.array(chain2_states)
        
        # Plot the chains
        plt.plot(chain1, 'b-', alpha=0.7, label='Chain 1 (starting from min)')
        plt.plot(chain2, 'r-', alpha=0.7, label='Chain 2 (starting from max)')
        
        # Highlight coalescence
        if coalescence_time < len(chain1) - 1:  # -1 because we include initial state
            plt.axvline(x=coalescence_time, color='g', linestyle='--', 
                     label=f'Coalescence at t={coalescence_time}')
        
        # Plot target density for reference
        ax2 = plt.twinx()
        x_grid = np.linspace(min(self.min_state, np.min(chain1)) - 1, 
                           max(self.max_state, np.max(chain2)) + 1, 1000)
        y_grid = np.array([self.target_density(x) for x in x_grid])
        y_grid = y_grid / np.max(y_grid)  # Normalize for visualization
        ax2.plot(x_grid, y_grid, 'k--', alpha=0.3, label='Target Density (normalized)')
        ax2.set_ylabel('Normalized Target Density')
        
        plt.title('Rudimentary Coupling Strategy for Perfect Sampling')
        plt.xlabel('Iteration')
        plt.ylabel('Parameter value')
        plt.legend(loc='upper right')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

    def plot_target_and_proposal(self, x_range: Optional[Tuple[float, float]] = None) -> None:
        """
        Plot the target density and proposal density to visualize the π/b ratio.
        
        Parameters:
        -----------
        x_range : tuple, optional
            Range of x values to plot (min, max)
        """
        if x_range is None:
            x_range = (self.min_state - 1, self.max_state + 1)
        
        x_grid = np.linspace(x_range[0], x_range[1], 1000)
        
        # Compute target density
        target_vals = np.array([self.target_density(x) for x in x_grid])
        
        # Compute proposal density (using middle point as reference)
        ref_point = (self.min_state + self.max_state) / 2
        proposal_vals = np.array([self.proposal_density(x, ref_point) for x in x_grid])
        
        # Compute π/b ratio
        ratio_vals = target_vals / proposal_vals
        
        # Create subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
        
        # Plot target and proposal densities
        ax1.plot(x_grid, target_vals, 'b-', label='Target π(x)')
        ax1.plot(x_grid, proposal_vals, 'r--', label='Proposal b(x|ref)')
        ax1.set_title('Target and Proposal Densities')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot π/b ratio
        ax2.plot(x_grid, ratio_vals, 'g-', label='π(x)/b(x|ref) Ratio')
        ax2.set_title('π/b Ratio (Basis for Ordering ≪)')
        ax2.set_xlabel('x')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Mark extremal points
        ax2.axvline(x=self.min_state, color='k', linestyle='--', label='Min State')
        ax2.axvline(x=self.max_state, color='k', linestyle=':', label='Max State')
        
        plt.tight_layout()
        plt.show()


# Example usage with a beta mixture distribution
if __name__ == "__main__":
    # Define a mixture of beta distributions as the target
    def target_density(x):
        """
        Unnormalized density of a mixture of beta distributions.
        This creates a bimodal distribution on [0, 1].
        """
        return 0.7 * stats.beta.pdf(x, 2, 5) + 0.3 * stats.beta.pdf(x, 8, 2)
    
    # Define a proposal function (e.g., truncated normal)
    def proposal_sampler(x):
        """Sample from a truncated normal centered at x."""
        proposal = x + np.random.normal(0, 0.15)
        # Ensure proposal is within [0, 1]
        return np.clip(proposal, 0, 1)
    
    def proposal_density(y, x):
        """
        Density of the proposal from x to y (truncated normal).
        This is a simplification since computing the exact density of a truncated normal
        requires normalization constants.
        """
        # Distance from x to y
        diff = y - x
        # Gaussian density proportional to exp(-diff²/(2σ²))
        return np.exp(-diff**2 / (2 * 0.15**2))
    
    # Find approximate extremal points for π/b
    def find_extremal_points():
        """Find approximate extremal points of π/b ratio."""
        x_grid = np.linspace(0, 1, 1000)
        ratio_vals = np.zeros_like(x_grid)
        
        # Compute π/b ratio at each point
        ref_point = 0.5  # Reference point for proposal
        for i, x in enumerate(x_grid):
            ratio_vals[i] = target_density(x) / proposal_density(x, ref_point)
        
        # Find min and max
        min_idx = np.argmin(ratio_vals)
        max_idx = np.argmax(ratio_vals)
        
        return x_grid[min_idx], x_grid[max_idx]
    
    # Compute extremal points
    min_state, max_state = find_extremal_points()
    print(f"Approximate extremal points for π/b ratio: min={min_state:.4f}, max={max_state:.4f}")
    
    # Initialize the sampler
    sampler = RudimentaryCouplingPerfectSampler(
        target_density=target_density,
        proposal_density=proposal_density,
        proposal_sampler=proposal_sampler,
        min_state=min_state,
        max_state=max_state,
        max_iterations=2000,
        seed=42
    )
    
    # Plot target and proposal to visualize ordering
    sampler.plot_target_and_proposal((0, 1))
    
    # Verify that the coupling preserves ordering
    sampler.verify_ordering_preservation(num_points=200)
    
    # Generate a perfect sample
    sample, coalescence_time, chain1_states, chain2_states = sampler.sample()
    print(f"Perfect sample: {sample:.4f}")
    
    # Visualize the chains
    sampler.visualize_chains(chain1_states, chain2_states, coalescence_time)
    
    # Generate multiple samples to check distribution
    n_samples = 500
    samples = []
    coalescence_times = []
    
    for i in range(n_samples):
        if i % 100 == 0:
            print(f"Generating sample {i}/{n_samples}")
        
        sample, t, _, _ = sampler.sample()
        samples.append(sample)
        coalescence_times.append(t)
    
    # Plot the resulting distribution
    plt.figure(figsize=(12, 8))
    
    # Create grid for true target density
    x_grid = np.linspace(0, 1, 1000)
    true_density = np.array([target_density(x) for x in x_grid])
    true_density = true_density / np.trapz(true_density, x_grid)  # Normalize
    
    # Plot histogram of samples and true density
    plt.subplot(2, 1, 1)
    sns.histplot(samples, stat='density', bins=30, alpha=0.6, label='Perfect samples')
    plt.plot(x_grid, true_density, 'r-', label='True density')
    plt.title('Distribution of Perfect Samples')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Plot histogram of coalescence times
    plt.subplot(2, 1, 2)
    sns.histplot(coalescence_times, bins=30, kde=True)
    plt.axvline(x=np.mean(coalescence_times), color='r', linestyle='--', 
               label=f'Mean: {np.mean(coalescence_times):.2f}')
    plt.title('Distribution of Coalescence Times')
    plt.xlabel('Coalescence Time')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

## Sample Path Analysis of CFTP Algorithm

Let's examine the fundamental structure of the Coupled From The Past (CFTP) algorithm through sample path analysis. The figure shows sample paths of both chains $(c_t^{(1)})$ and $(c_t^{(2)})$ for a successful CFTP run for $n$-component mixture of normals ($n = 25$ and $m = 25$).

![Sample paths of CFTP chains](cftp_sample_paths.png)

*Fig. 13.5. Sample path of both chains $(c_t^{(1)})$ and $(c_t^{(2)})$ for a successful CFTP on $n$-component mixture of normals ($n = 25$ and $m = 25$).*

The above example shows the fundamental structure of this automatic CFTP algorithm: it relies on a single chain, since the second chain is only used through its starting value. When the simulated proposals for the two chains merge, the simulated value from $1$ satisfies:

$$\frac{\pi(c_t^{(1)})}{b(c_t^{(1)}|y)} \leq \frac{\pi(y)}{b(y|c_t^{(1)})}$$

It is therefore exactly distributed from $\pi$ since this is the usual Accept-Reject step. Therefore, in the case of independent proposals, the automatic CFPT algorithm is equivalent to Accept-Reject when there are coalescences. (See Problem 13.15 for more interesting extensions). We now consider a second and much less obvious case of an Accept-Reject scheme corresponding to a perfect sampling algorithm.

## 13.3 Forward Coupling

Despite the generic complexity of backward iterations, which can require an exponential or possibly even an infinite number of backward iterations when the chain is uniformly ergodic (see Section 13.3.3), there are still problems with the CFTP algorithm in that the running time is sometimes potentially unbounded. Consider the practical case of a user who cannot wait for a simulation to come out and interrupts a CFTP run, thus obtaining a sample from the wrong distribution. This abortion of long runs modifies the output of the algorithm and does not produce exact simulations from $\pi$ (Problem 13.16). To alleviate this difficulty, Fill (1998) introduced a modification of CFTP which can be interrupted while preserving the central feature of CFTP.

CFTP algorithm tries to reproduce the effect of running $(c_t)$ from time $-\infty$ to 0, where $-\infty$ stands for a "time long enough in the past", so that when selecting the value at time 0, $c_0$, this value is exactly distributed according to the stationary distribution.

## Mathematical Formulation of Forward Coupling

The concept of forward coupling addresses the limitations of the CFTP algorithm by modifying how we approach the simulation process. Instead of starting from the past and running forward, we can consider an alternative approach.

Let us denote by $\{X_t\}$ the Markov chain of interest with transition kernel $P$ and stationary distribution $\pi$. The standard CFTP algorithm starts multiple chains from different initial states at time $-T$ and runs them forward until time 0, checking for coalescence. If coalescence occurs, the common value at time 0 is a sample from $\pi$.

In forward coupling, we take a different approach:

1. We start a chain $\{X_t\}$ from an arbitrary initial state $X_0 = x_0$
2. Simultaneously, we run an "independent" chain $\{Y_t\}$ where each state $Y_t$ is sampled directly from $\pi$
3. We construct a coupling between these two chains such that once they meet (coalesce), they remain identical thereafter

The key advantage is that we can define a stopping time $\tau = \min\{t \geq 0: X_t = Y_t\}$, and for any time $T \geq \tau$, the value $X_T$ is exactly distributed according to $\pi$.

## The Forward Coupling Algorithm

The forward coupling algorithm can be summarized as follows:

1. Initialize $X_0 = x_0$ (arbitrary starting point)
2. For $t = 0, 1, 2, \ldots$:
   a. Sample $Y_t \sim \pi$ (directly from the target distribution)
   b. With probability $\alpha(X_t, Y_t)$, set $X_{t+1} = Y_t$ (coupling occurs)
   c. Otherwise, generate $X_{t+1}$ according to the transition kernel $P$
   d. If $X_{t+1} = Y_t$, set $\tau = t+1$ and stop

3. Return $X_\tau$ as a perfect sample from $\pi$

The coupling probability $\alpha(X_t, Y_t)$ is chosen to maximize the chance of coalescence while maintaining the correct dynamics of the individual chains.

## Advantages of Forward Coupling

The main advantages of forward coupling over CFTP include:

1. **Interruptible Nature**: The algorithm can be interrupted at any time, and we can still obtain a sample from $\pi$ by continuing to run the coupled chains until they meet.

2. **Bounded Expected Running Time**: Under appropriate conditions, the expected running time is bounded, which addresses the potential unboundedness issue in CFTP.

3. **Simplified Implementation**: The forward-time nature of the algorithm makes it conceptually simpler to implement and understand.

4. **Monitoring Progress**: It's easier to monitor progress towards coalescence, as we don't need to restart simulations from further in the past.

## Mathematical Properties

The correctness of forward coupling relies on the following key properties:

1. **Marginal Correctness**: Each chain maintains its correct marginal distribution. $X_t$ follows the dynamics of the Markov chain starting from $x_0$, while $Y_t \sim \pi$ at all times.

2. **Coupling Property**: Once the chains meet, they remain identical thereafter, i.e., $X_\tau = Y_\tau$ implies $X_t = Y_t$ for all $t \geq \tau$.

3. **Coupling Time Distribution**: The distribution of the coupling time $\tau$ is related to the mixing properties of the Markov chain.

The expected coupling time $\mathbb{E}[\tau]$ is bounded when the chain is uniformly ergodic, which ensures practical efficiency of the algorithm.





## Example Implementation Considerations

When implementing forward coupling:

1. **Efficient Sampling from π**: The algorithm requires direct sampling from $\pi$, which might not be feasible in general. However, in certain cases (like when $\pi$ is a discrete distribution with small state space), this can be practical.

2. **Coupling Mechanism**: The choice of coupling mechanism affects the efficiency of the algorithm. Maximal couplings minimize the expected coupling time but might be difficult to implement.

3. **Monitoring Convergence**: Unlike CFTP, we don't need to explicitly check for coalescence across multiple chains - we only need to detect when $X_t = Y_t$.

## Conclusion

Forward coupling provides an elegant solution to the potential unboundedness of CFTP runtime while maintaining the perfect sampling property. By allowing interruptible execution, it makes perfect sampling more practical for complex problems where simulation times might be prohibitively long.

While it requires the ability to sample directly from $\pi$ (which CFTP doesn't), this tradeoff can be worthwhile in scenarios where runtime predictability is crucial or when the CFTP algorithm's backward coupling structure is difficult to implement.



```python
import random

class ExponentialMixtureCFTP:
    def __init__(self, n, n0):
        self.n = n
        self.n0 = n0
        self.state_space = range(2)  # Two-component mixture

    def transition(self, state, random_value):
        """
        Simulates a transition of the Markov chain.

        Args:
            state (int): The current state (0 or 1).
            random_value (float): A uniform random value in [0, 1).

        Returns:
            int: The next state (0 or 1).
        """
        if state == 0:
            if random_value < 0.5:  # Example transition probability
                return 0
            else:
                return 1
        elif state == 1:
            if random_value < 0.7:  # Example transition probability
                return 1
            else:
                return 0
        return -1 # Should not happen

    def run_backward(self, initial_states, random_sequence):
        """
        Runs the Markov chain backward in time from a set of initial states.

        Args:
            initial_states (list): A list of initial states.
            random_sequence (list): A sequence of uniform random values for the backward steps.

        Returns:
            list: The states at time 0 for each initial state.
        """
        current_states = list(initial_states)
        for rand_val in reversed(random_sequence):
            next_states = []
            for state in current_states:
                # To go backward, we need the inverse of the transition.
                # This depends on the specific transition probabilities.
                # For this example, let's assume a simplified inverse.
                if state == 0:
                    if rand_val < 0.5:
                        next_states.append(0)
                    else:
                        next_states.append(1)
                elif state == 1:
                    if rand_val < 0.7:
                        next_states.append(1)
                    else:
                        next_states.append(0)
            current_states = next_states
        return current_states

    def sample(self, max_backward_steps=20):
        """
        Implements the Coupling From The Past (CFTP) algorithm.

        Args:
            max_backward_steps (int): The maximum number of backward steps to consider.

        Returns:
            int or None: A sample from the stationary distribution, or None if coalescence
                         is not achieved within the maximum steps.
        """
        t = 1
        while t <= max_backward_steps:
            random_sequence = [random.random() for _ in range(t)]
            initial_states = list(self.state_space)
            final_states = self.run_backward(initial_states, random_sequence)

            if all(s == final_states[0] for s in final_states):
                return final_states[0]
            t *= 2  # Double the backward steps

        return None

if __name__ == "__main__":
    n_samples = 10
    n = 81
    n0 = 59
    cftp_sampler = ExponentialMixtureCFTP(n, n0)

    samples = [cftp_sampler.sample() for _ in range(n_samples)]
    print(f"Generated {n_samples} samples using CFTP:")
    print(samples)

    if all(sample is not None for sample in samples):
        counts = {0: samples.count(0), 1: samples.count(1)}
        print("Sample counts:", counts)
```

# 13.6.2 Perfect Sampling and Tempering

Tempering, as introduced in Marinari and Parisi (1992), Geyer and Thompson (1995), and Neal (1996), appears as an augmented annealing, in that the distribution is taken over a set of $k$ scales. From an analytical perspective, choices are a sequence $(π_n)$ of target distributions, with the "hotter" distributions enabling flow here and there and less concentrated. The purpose behind this scheme is to facilitate the moves of a Metropolis–Hastings sampler by somehow flattening the modes of the target distribution. This device is, for instance, used in Celeux et al. (2000) to impose a proper level of label switching in the simulation of the posterior distribution associated with a mixture of distributions.

Following the description of Meng and Wong (2001), a sequence of target distributions $(π_n)_{1≤n≤k}$ is thus chosen such that $π_k = π$, $π_1$ is a well-known distribution for which simulation is possible, and $π_n$ is a distribution on a single atom, denoted B. Each of the $π_n$'s is defined on a possibly different space, $Θ_n$. Using the arbitrary values $θ^{(n)}$ in $Θ_n$ ($n = 1, \ldots, N^*$) and setting $θ = (θ^{(1)}, \ldots, θ^{(N^*)})$, the joint distribution

$$p(θ, n) = π_n(θ^{(n)}),$$

$\sum_{n=1}^{N^*} π_n = 1$,

a reversible jump (Section 11.2) algorithm can be constructed for this joint distribution. This is a Markov chain MC algorithm in $(θ, n)$ space, where $n$ is called a "level," with only $n$ changing. If $p_{n,n+1}$ and $p_{n,n-1}$ are the probabilities of moving up and down, respectively, from level $n$, $q_{n→n+1}$ is the probability of choosing a level-down move, and if $q_{n→n+1}(θ^{(n)})$ and $q_{n→n-1}(θ^{(n)})$ denote the proposal distributions for the corresponding moves, we thus assume that the usual symmetry condition (11.10) holds for $q_{n→n+1}$ and $q_{n+1→n}$, getting

$$ρ_{n→n+1}(θ', θ) = \frac{π_{n+1}(θ'^{(n+1)})p_{n+1,n}q_{n+1→n}(θ'^{(n+1)})}{π_n(θ^{(n)})p_{n,n+1}q_{n→n+1}(θ^{(n)})} \times A_1$$

If the acceptance probability of an upward move is $ρ_{n→n+1}(θ', θ)$, the symmetry translates into

$$ρ_{n+1→n}(θ', θ) = 1/ρ_{n→n+1}(θ, θ')$$

as the acceptance probability of a downward move.

### Example 13.16: Power tempering
Assuming a compact space like $[0, 1]^p$, a typical characteristic of tempering is to define a sequence of powers $κ_1 > \cdots > κ_k = 1$, such that

$$π_n \propto π^{κ_n}.$$

These distributions are literally flattened versions of the target distribution $π$. The corresponding acceptance probability for a move up from n to n + 1 is then proportional to

$$\frac{π_{n+1}(θ'^{(n+1)})}{π_n(θ^{(n)})} = \frac{π(θ'^{(n+1)})^{κ_{n+1}}}{π(θ^{(n)})^{κ_n}} = π(θ'^{(n+1)})^{κ_{n+1} - κ_n}$$

Green and Mira (1999) address a further difficulty by designing an equal number of moves from n to n + 1 and from n + 1 to n, and by accepting the entire sequence as a single proposal, thus canceling the normalizing constants in the acceptance probability.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random

class PowerTempering:
    """
    Implementation of Power Tempering as described in section 13.6.2 
    (Perfect Sampling and Tempering).
    
    This implementation uses the concept of tempering with a sequence of
    distributions π_n that are powered versions of the target distribution π,
    with π_n ∝ π^κ_n where κ_1 > ... > κ_k = 1.
    """
    
    def __init__(self, target_dist, kappas, proposal_dist=None):
        """
        Initialize the Power Tempering sampler.
        
        Parameters:
        -----------
        target_dist : callable
            Function that evaluates the unnormalized target distribution π(θ)
        kappas : list
            Sequence of tempering powers κ_1 > ... > κ_k = 1
        proposal_dist : callable, optional
            Function that generates proposals. If None, a default normal proposal is used
        """
        self.target_dist = target_dist
        self.kappas = sorted(kappas, reverse=True)  # Ensure κ_1 > ... > κ_k = 1
        
        # Ensure the last kappa is 1 (targeting the actual distribution)
        if self.kappas[-1] != 1:
            print("Warning: Last kappa should be 1. Setting κ_k = 1.")
            self.kappas[-1] = 1
            
        self.num_levels = len(kappas)
        
        # Default proposal is a normal distribution
        if proposal_dist is None:
            self.proposal_dist = lambda x: np.random.normal(loc=x, scale=0.1)
        else:
            self.proposal_dist = proposal_dist
    
    def _tempered_dist(self, theta, level):
        """
        Evaluate the tempered distribution π_n(θ) ∝ π(θ)^κ_n
        
        Parameters:
        -----------
        theta : float or array
            Parameter value to evaluate
        level : int
            Level index n (0-based) for which to compute π_n
            
        Returns:
        --------
        float
            The unnormalized density evaluation
        """
        # Compute π(θ)^κ_n
        base_density = self.target_dist(theta)
        return base_density ** self.kappas[level]
    
    def level_transition_prob(self, current_theta, current_level, proposed_level):
        """
        Calculate the acceptance probability for changing levels
        
        Parameters:
        -----------
        current_theta : float or array
            Current parameter value
        current_level : int
            Current level index (0-based)
        proposed_level : int
            Proposed level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the level change
        """
        # Calculate the acceptance ratio for moving between levels
        if proposed_level > current_level:  # Moving up
            # Acceptance ratio: (π_{n+1}(θ)/π_n(θ)) = π(θ)^(κ_{n+1} - κ_n)
            kappa_diff = self.kappas[proposed_level] - self.kappas[current_level]
            base_density = self.target_dist(current_theta)
            return min(1, base_density ** kappa_diff)
        else:  # Moving down
            # Reverse of moving up: 1/ρ_{n→n+1}(θ', θ)
            kappa_diff = self.kappas[current_level] - self.kappas[proposed_level]
            base_density = self.target_dist(current_theta)
            return min(1, base_density ** (-kappa_diff))
    
    def within_level_transition_prob(self, current_theta, proposed_theta, level):
        """
        Calculate the acceptance probability for within-level moves
        
        Parameters:
        -----------
        current_theta : float or array
            Current parameter value
        proposed_theta : float or array
            Proposed parameter value
        level : int
            Current level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the parameter change within the level
        """
        # Standard Metropolis-Hastings ratio for tempered distribution
        current_density = self._tempered_dist(current_theta, level)
        proposed_density = self._tempered_dist(proposed_theta, level)
        
        return min(1, proposed_density / current_density)
    
    def sample(self, n_samples, burn_in=1000, thin=1):
        """
        Generate samples using the Power Tempering algorithm
        
        Parameters:
        -----------
        n_samples : int
            Number of samples to generate
        burn_in : int, optional
            Number of samples to discard as burn-in
        thin : int, optional
            Thinning factor (keep every 'thin' samples)
            
        Returns:
        --------
        list
            Samples from the target distribution π
        """
        # Initial state
        current_theta = 0.5  # Starting point (could be random)
        current_level = 0    # Start at the highest temperature (most flattened)
        
        # Storage for samples (we only keep samples from level k where κ_k = 1)
        samples = []
        
        # Total iterations needed
        total_iterations = burn_in + n_samples * thin
        
        # Counters for diagnostics
        accepted_level_moves = 0
        proposed_level_moves = 0
        accepted_theta_moves = 0
        proposed_theta_moves = 0
        
        for i in range(total_iterations):
            # Decide whether to propose a level change or parameter change
            if np.random.random() < 0.5:  # 50% chance to change level
                # Propose a new level (up or down one level)
                proposed_level_moves += 1
                if current_level == 0:  # At the highest level, can only go down
                    proposed_level = 1
                elif current_level == self.num_levels - 1:  # At the lowest level, can only go up
                    proposed_level = self.num_levels - 2
                else:  # Can go either up or down
                    proposed_level = current_level + np.random.choice([-1, 1])
                
                # Calculate acceptance probability
                accept_prob = self.level_transition_prob(current_theta, current_level, proposed_level)
                
                # Accept or reject the level move
                if np.random.random() < accept_prob:
                    current_level = proposed_level
                    accepted_level_moves += 1
            else:  # 50% chance to change parameter value
                # Propose a new parameter value
                proposed_theta_moves += 1
                proposed_theta = self.proposal_dist(current_theta)
                
                # Calculate acceptance probability
                accept_prob = self.within_level_transition_prob(current_theta, proposed_theta, current_level)
                
                # Accept or reject the parameter move
                if np.random.random() < accept_prob:
                    current_theta = proposed_theta
                    accepted_theta_moves += 1
            
            # Store sample if past burn-in and at the right thinning interval
            if i >= burn_in and (i - burn_in) % thin == 0:
                # Only store samples from the target distribution (level k where κ_k = 1)
                if current_level == self.num_levels - 1:
                    samples.append(current_theta)
        
        # Print diagnostics
        print(f"Level moves: {accepted_level_moves}/{proposed_level_moves} "
              f"({accepted_level_moves/proposed_level_moves:.2%} accepted)")
        print(f"Parameter moves: {accepted_theta_moves}/{proposed_theta_moves} "
              f"({accepted_theta_moves/proposed_theta_moves:.2%} accepted)")
        
        return np.array(samples)


# Example usage with a bimodal target distribution
def demo_power_tempering():
    """
    Demonstrate the Power Tempering algorithm on a bimodal distribution.
    """
    # Define a bimodal target distribution (mixture of two Gaussians)
    def bimodal_target(x):
        # Mixture of two Gaussians at -2 and 2 with standard deviation 0.5
        comp1 = stats.norm.pdf(x, loc=-2, scale=0.5)
        comp2 = stats.norm.pdf(x, loc=2, scale=0.5)
        return 0.5 * comp1 + 0.5 * comp2
    
    # Define a custom proposal distribution
    def proposal(x):
        return x + np.random.normal(0, 0.5)
    
    # Define a sequence of tempering powers
    kappas = [0.1, 0.3, 0.6, 1.0]  # κ_1 > ... > κ_4 = 1
    
    # Create the tempering sampler
    sampler = PowerTempering(
        target_dist=bimodal_target,
        kappas=kappas,
        proposal_dist=proposal
    )
    
    # Generate samples
    samples = sampler.sample(n_samples=5000, burn_in=1000, thin=2)
    
    # Plot the results
    plt.figure(figsize=(12, 6))
    
    # Plot the histogram of samples
    plt.subplot(1, 2, 1)
    plt.hist(samples, bins=50, density=True, alpha=0.7, label='MCMC Samples')
    
    # Plot the true target density
    x = np.linspace(-4, 4, 1000)
    true_density = [bimodal_target(xi) for xi in x]
    # Normalize the density (since our target is unnormalized)
    true_density = np.array(true_density) / np.trapz(true_density, x)
    plt.plot(x, true_density, 'r-', linewidth=2, label='True Density')
    
    plt.title('Power Tempering MCMC')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    # Plot the tempered distributions
    plt.subplot(1, 2, 2)
    for i, kappa in enumerate(kappas):
        # Plot each tempered version of the distribution
        tempered_density = [bimodal_target(xi)**kappa for xi in x]
        # Normalize for visualization
        tempered_density = np.array(tempered_density) / np.trapz(tempered_density, x)
        plt.plot(x, tempered_density, label=f'κ_{i+1}={kappa}')
    
    plt.title('Tempered Distributions')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    demo_power_tempering()
```

# Augmented Chains and Dominating Processes

Since the augmented chain $(θ_t, N_t)$ is associated with the joint distribution $p(θ,n) = π_n(θ^{(n)})$, monitoring creates many iterations of the MCMC sampler. Therefore, monitoring the output of the MCMC sampler for only the times when $N_t$ produces a subchain converging to the target distribution $π$ appears like a costly device, since the simulations corresponding to the other values of $N_t$ are lost. But the potential gain is in an enhanced exploration of the state space $Θ_K$. Note also that the other simulations can sometimes be recycled in importance sampling estimators.

There exists a large collection of references in the MCMC literature (Geyer and Thompson 1995, Müller 1999, Celeux et al. 2000). What is of most interest in connection with this chapter is that it is fairly easy to define a dominating process on $(0, 1, \ldots, N^*)$ as shown by Meller and Nicholls (2001). To achieve this domination, they assume in addition that

$$ρ_{n→n+1}(θ,θ′) \leq κ_n \frac{π_{n+1}}{π_n},$$

that is, that the upward and downward transitions are bounded away from 0. The dominating process $D_t$ is then associated with the same proposal as $N_t$ and the acceptance probabilities

$$α_{n→n+1} = \min\left(1, κ_n \frac{π_{n+1}}{π_n}\right),$$

$$α_{n→n-1} = \min\left(1, \frac{π_{n-1}}{κ_{n-1}π_{n}}\right).$$

We couple $N_t$ and $D_t$ so that (i) the upward, downward, and fixed-level moves are chosen with the same probabilities, and (ii) the acceptance is based on the same uniform. Then:

## Lemma 13.17
*The chain $(D_t)$ dominates the chain $(N_t)$ in that*

$$\mathbb{P}(N_t = n | N_0 = D_0 = n) \leq \mathbb{P}(D_t = n | N_0 = D_0 = n)$$   *when*   $n < m$.

This result obviously holds by construction of the acceptance probabilities $α_{n→n+1}$ and $α_{n→n-1}$, since those favor upward moves against downward moves. Note that $(D_t)$ is necessarily ergodic, as a Markov chain on a finite state space, and

$$q_n \propto κ_n \frac{π_{n+1}}{π_n}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random

class PowerTemperingWithDomination:
    """
    Implementation of Power Tempering with a dominating process as described in the text.
    This implementation builds on the concept of tempering with a sequence of
    distributions π_n that are powered versions of the target distribution π,
    with π_n ∝ π^κ_n where κ_1 > ... > κ_k = 1, and adds a dominating process.
    """
    
    def __init__(self, target_dist, kappas, proposal_dist=None, coupling_constants=None):
        """
        Initialize the Power Tempering sampler with dominating process.
        
        Parameters:
        -----------
        target_dist : callable
            Function that evaluates the unnormalized target distribution π(θ)
        kappas : list
            Sequence of tempering powers κ_1 > ... > κ_k = 1
        proposal_dist : callable, optional
            Function that generates proposals. If None, a default normal proposal is used
        coupling_constants : list, optional
            Constants κ_n used for bounding the transitions. If None, computed automatically
        """
        self.target_dist = target_dist
        self.kappas = sorted(kappas, reverse=True)  # Ensure κ_1 > ... > κ_k = 1
        
        # Ensure the last kappa is 1 (targeting the actual distribution)
        if self.kappas[-1] != 1:
            print("Warning: Last kappa should be 1. Setting κ_k = 1.")
            self.kappas[-1] = 1
            
        self.num_levels = len(kappas)
        
        # Default proposal is a normal distribution
        if proposal_dist is None:
            self.proposal_dist = lambda x: np.random.normal(loc=x, scale=0.1)
        else:
            self.proposal_dist = proposal_dist
            
        # Coupling constants for dominating process
        if coupling_constants is None:
            # By default, we use a simple heuristic to set the coupling constants
            self.coupling_constants = [0.7] * (self.num_levels - 1)
        else:
            if len(coupling_constants) != self.num_levels - 1:
                raise ValueError("coupling_constants should have length num_levels - 1")
            self.coupling_constants = coupling_constants
            
        # Pre-compute ratios of consecutive temperatures for efficiency
        self.pi_ratios = []
        for i in range(self.num_levels - 1):
            self.pi_ratios.append(self.kappas[i+1] / self.kappas[i])
    
    def _tempered_dist(self, theta, level):
        """
        Evaluate the tempered distribution π_n(θ) ∝ π(θ)^κ_n
        
        Parameters:
        -----------
        theta : float or array
            Parameter value to evaluate
        level : int
            Level index n (0-based) for which to compute π_n
            
        Returns:
        --------
        float
            The unnormalized density evaluation
        """
        # Compute π(θ)^κ_n
        base_density = self.target_dist(theta)
        return base_density ** self.kappas[level]
    
    def level_transition_prob(self, current_theta, current_level, proposed_level):
        """
        Calculate the acceptance probability for changing levels
        
        Parameters:
        -----------
        current_theta : float or array
            Current parameter value
        current_level : int
            Current level index (0-based)
        proposed_level : int
            Proposed level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the level change
        """
        # Calculate the acceptance ratio for moving between levels
        if proposed_level > current_level:  # Moving up
            # Acceptance ratio: min(1, π_{n+1}(θ)/π_n(θ)) = min(1, π(θ)^(κ_{n+1} - κ_n))
            kappa_diff = self.kappas[proposed_level] - self.kappas[current_level]
            base_density = self.target_dist(current_theta)
            return min(1, base_density ** kappa_diff)
        else:  # Moving down
            # Reverse of moving up
            kappa_diff = self.kappas[current_level] - self.kappas[proposed_level]
            base_density = self.target_dist(current_theta)
            return min(1, base_density ** (-kappa_diff))
    
    def dominating_level_transition_prob(self, current_level, proposed_level):
        """
        Calculate the acceptance probability for the dominating process when changing levels.
        This implementation uses the bounds from the paper:
        α_{n→n+1} = min(1, κ_n * π_{n+1}/π_n)
        α_{n→n-1} = min(1, π_{n-1}/(κ_{n-1} * π_n))
        
        Parameters:
        -----------
        current_level : int
            Current level index (0-based)
        proposed_level : int
            Proposed level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the level change in the dominating process
        """
        if proposed_level > current_level:  # Moving up
            # We're using pre-computed ratios of consecutive temperature levels
            k = self.coupling_constants[current_level]
            pi_ratio = self.pi_ratios[current_level]
            return min(1, k * pi_ratio)
        else:  # Moving down
            # We're using pre-computed ratios of consecutive temperature levels
            k = self.coupling_constants[proposed_level]
            pi_ratio = self.pi_ratios[proposed_level]
            return min(1, 1 / (k * pi_ratio))
    
    def within_level_transition_prob(self, current_theta, proposed_theta, level):
        """
        Calculate the acceptance probability for within-level moves
        
        Parameters:
        -----------
        current_theta : float or array
            Current parameter value
        proposed_theta : float or array
            Proposed parameter value
        level : int
            Current level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the parameter change within the level
        """
        # Standard Metropolis-Hastings ratio for tempered distribution
        current_density = self._tempered_dist(current_theta, level)
        proposed_density = self._tempered_dist(proposed_theta, level)
        
        return min(1, proposed_density / current_density)
    
    def sample_with_domination(self, n_samples, burn_in=1000, thin=1):
        """
        Generate samples using the Power Tempering algorithm with a coupled dominating process
        
        Parameters:
        -----------
        n_samples : int
            Number of samples to generate
        burn_in : int, optional
            Number of samples to discard as burn-in
        thin : int, optional
            Thinning factor (keep every 'thin' samples)
            
        Returns:
        --------
        tuple
            (samples, dominating_levels, actual_levels) where:
            - samples are from the target distribution π
            - dominating_levels tracks the levels of the dominating process
            - actual_levels tracks the levels of the actual process
        """
        # Initial state
        current_theta = 0.5  # Starting point (could be random)
        current_level = 0    # Start at the highest temperature (most flattened)
        
        # Initialize both processes at the same level
        dominating_level = current_level
        
        # Storage for samples (we only keep samples from level k where κ_k = 1)
        samples = []
        dominating_levels = []
        actual_levels = []
        
        # Total iterations needed
        total_iterations = burn_in + n_samples * thin
        
        # Counters for diagnostics
        accepted_level_moves_actual = 0
        accepted_level_moves_dominating = 0
        proposed_level_moves = 0
        accepted_theta_moves = 0
        proposed_theta_moves = 0
        
        for i in range(total_iterations):
            # Decide whether to propose a level change or parameter change
            if np.random.random() < 0.5:  # 50% chance to change level
                # Propose a new level (up or down one level)
                proposed_level_moves += 1
                if current_level == 0:  # At the highest level, can only go down
                    proposed_level = 1
                elif current_level == self.num_levels - 1:  # At the lowest level, can only go up
                    proposed_level = self.num_levels - 2
                else:  # Can go either up or down
                    proposed_level = current_level + np.random.choice([-1, 1])
                
                # Generate a uniform random variable for coupled acceptance
                u = np.random.random()
                
                # Calculate acceptance probabilities for both processes
                accept_prob_actual = self.level_transition_prob(
                    current_theta, current_level, proposed_level)
                accept_prob_dominating = self.dominating_level_transition_prob(
                    dominating_level, proposed_level)
                
                # Accept or reject the level move for both processes using the same uniform
                if u < accept_prob_dominating:
                    dominating_level = proposed_level
                    accepted_level_moves_dominating += 1
                    
                    # The actual process can only move if the dominating process moves
                    # AND its own acceptance criterion is met
                    if u < accept_prob_actual:
                        current_level = proposed_level
                        accepted_level_moves_actual += 1
                
            else:  # 50% chance to change parameter value
                # Propose a new parameter value
                proposed_theta_moves += 1
                proposed_theta = self.proposal_dist(current_theta)
                
                # Calculate acceptance probability
                accept_prob = self.within_level_transition_prob(
                    current_theta, proposed_theta, current_level)
                
                # Accept or reject the parameter move
                if np.random.random() < accept_prob:
                    current_theta = proposed_theta
                    accepted_theta_moves += 1
            
            # Store sample if past burn-in and at the right thinning interval
            if i >= burn_in and (i - burn_in) % thin == 0:
                # Store the levels for both processes
                dominating_levels.append(dominating_level)
                actual_levels.append(current_level)
                
                # Only store samples from the target distribution (level k where κ_k = 1)
                if current_level == self.num_levels - 1:
                    samples.append(current_theta)
        
        # Print diagnostics
        print(f"Actual level moves: {accepted_level_moves_actual}/{proposed_level_moves} "
              f"({accepted_level_moves_actual/proposed_level_moves:.2%} accepted)")
        print(f"Dominating level moves: {accepted_level_moves_dominating}/{proposed_level_moves} "
              f"({accepted_level_moves_dominating/proposed_level_moves:.2%} accepted)")
        print(f"Parameter moves: {accepted_theta_moves}/{proposed_theta_moves} "
              f"({accepted_theta_moves/proposed_theta_moves:.2%} accepted)")
        
        return np.array(samples), dominating_levels, actual_levels
    
    def perfect_sampling(self, max_iterations=10000):
        """
        Implement perfect sampling using the dominating process.
        
        This implements the technique described in the paper where the dominating process
        is used to achieve perfect sampling from the target distribution.
        
        Parameters:
        -----------
        max_iterations : int, optional
            Maximum number of iterations to attempt
        
        Returns:
        --------
        tuple
            (sample, iterations) where:
            - sample is a perfect sample from the target distribution
            - iterations is the number of iterations required
        """
        # Implementation of the perfect sampling algorithm would go here
        # This is a placeholder as the perfect sampling algorithm is not fully described
        # in the provided text snippet
        
        print("Perfect sampling is not fully implemented in this example.")
        return None, 0


# Example usage with a bimodal target distribution
def demo_power_tempering_with_domination():
    """
    Demonstrate the Power Tempering algorithm with dominating process on a bimodal distribution.
    """
    # Define a bimodal target distribution (mixture of two Gaussians)
    def bimodal_target(x):
        # Mixture of two Gaussians at -2 and 2 with standard deviation 0.5
        comp1 = stats.norm.pdf(x, loc=-2, scale=0.5)
        comp2 = stats.norm.pdf(x, loc=2, scale=0.5)
        return 0.5 * comp1 + 0.5 * comp2
    
    # Define a custom proposal distribution
    def proposal(x):
        return x + np.random.normal(0, 0.5)
    
    # Define a sequence of tempering powers
    kappas = [0.1, 0.3, 0.6, 1.0]  # κ_1 > ... > κ_4 = 1
    
    # Define coupling constants for the dominating process
    coupling_constants = [0.8, 0.9, 0.95]  # One fewer than the number of levels
    
    # Create the tempering sampler with dominating process
    sampler = PowerTemperingWithDomination(
        target_dist=bimodal_target,
        kappas=kappas,
        proposal_dist=proposal,
        coupling_constants=coupling_constants
    )
    
    # Generate samples with the coupled dominating process
    samples, dom_levels, act_levels = sampler.sample_with_domination(
        n_samples=5000, burn_in=1000, thin=2)
    
    # Plot the results
    plt.figure(figsize=(15, 10))
    
    # Plot the histogram of samples
    plt.subplot(2, 2, 1)
    plt.hist(samples, bins=50, density=True, alpha=0.7, label='MCMC Samples')
    
    # Plot the true target density
    x = np.linspace(-4, 4, 1000)
    true_density = [bimodal_target(xi) for xi in x]
    # Normalize the density (since our target is unnormalized)
    true_density = np.array(true_density) / np.trapz(true_density, x)
    plt.plot(x, true_density, 'r-', linewidth=2, label='True Density')
    
    plt.title('Power Tempering MCMC with Dominating Process')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    # Plot the tempered distributions
    plt.subplot(2, 2, 2)
    for i, kappa in enumerate(kappas):
        # Plot each tempered version of the distribution
        tempered_density = [bimodal_target(xi)**kappa for xi in x]
        # Normalize for visualization
        tempered_density = np.array(tempered_density) / np.trapz(tempered_density, x)
        plt.plot(x, tempered_density, label=f'κ_{i+1}={kappa}')
    
    plt.title('Tempered Distributions')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    # Plot the level trajectories
    plt.subplot(2, 2, 3)
    # Only plot a subset of the iterations for clarity
    n_plot = min(1000, len(dom_levels))
    iterations = range(n_plot)
    plt.plot(iterations, dom_levels[:n_plot], 'b-', alpha=0.6, label='Dominating Process')
    plt.plot(iterations, act_levels[:n_plot], 'r-', alpha=0.6, label='Actual Process')
    
    plt.title('Level Trajectories')
    plt.xlabel('Iteration')
    plt.ylabel('Level')
    plt.legend()
    
    # Plot the distribution of levels
    plt.subplot(2, 2, 4)
    plt.hist(dom_levels, bins=np.arange(len(kappas)+1)-0.5, alpha=0.6, label='Dominating Process')
    plt.hist(act_levels, bins=np.arange(len(kappas)+1)-0.5, alpha=0.6, label='Actual Process')
    
    plt.title('Distribution of Levels')
    plt.xlabel('Level')
    plt.ylabel('Frequency')
    plt.xticks(range(len(kappas)))
    plt.legend()
    
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    demo_power_tempering_with_domination()
```

            k = self.coupling_constants[proposed_level]

            return min(1, 1 / (k * pi_ratio))
    
    def within_level_transition_prob(self, current_theta, proposed_theta, level):
        """
        Calculate the acceptance probability for within-level moves
        
        Parameters:
        -----------
        current_theta : float or array
            Current parameter value
        proposed_theta : float or array
            Proposed parameter value
        level : int
            Current level index (0-based)
            
        Returns:
        --------
        float
            Acceptance probability for the parameter change within the level
        """
        # Standard Metropolis-Hastings ratio for tempered distribution
        current_density = self._tempered_dist(current_theta, level)
        proposed_density = self._tempered_dist(proposed_theta, level)
        
        return min(1, proposed_density / current_density)
    
    def sample_with_domination(self, n_samples, burn_in=1000, thin=1):
        """
        Generate samples using the Power Tempering algorithm with a coupled dominating process
        
        Parameters:
        -----------
        n_samples : int
            Number of samples to generate
        burn_in : int, optional
            Number of samples to discard as burn-in
        thin : int, optional
            Thinning factor (keep every 'thin' samples)
            
        Returns:
        --------
        tuple
            (samples, dominating_levels, actual_levels) where:
            - samples are from the target distribution π
            - dominating_levels tracks the levels of the dominating process
            - actual_levels tracks the levels of the actual process
        """
        # Initial state
        current_theta = 0.5  # Starting point (could be random)
        current_level = 0    # Start at the highest temperature (most flattened)
        
        # Initialize both processes at the same level
        dominating_level = current_level
        
        # Storage for samples (we only keep samples from level k where κ_k = 1)
        samples = []
        dominating_levels = []
        actual_levels = []
        
        # Total iterations needed
        total_iterations = burn_in + n_samples * thin
        
        # Counters for diagnostics
        accepted_level_moves_actual = 0
        accepted_level_moves_dominating = 0
        proposed_level_moves = 0
        accepted_theta_moves = 0
        proposed_theta_moves = 0
        
        for i in range(total_iterations):
            # Decide whether to propose a level change or parameter change
            if np.random.random() < 0.5:  # 50% chance to change level
                # Propose a new level (up or down one level)
                proposed_level_moves += 1
                if current_level == 0:  # At the highest level, can only go down
                    proposed_level = 1
                elif current_level == self.num_levels - 1:  # At the lowest level, can only go up
                    proposed_level = self.num_levels - 2
                else:  # Can go either up or down
                    proposed_level = current_level + np.random.choice([-1, 1])
                
                # Generate a uniform random variable for coupled acceptance
                u = np.random.random()
                
                # Calculate acceptance probabilities for both processes
                accept_prob_actual = self.level_transition_prob(
                    current_theta, current_level, proposed_level)
                accept_prob_dominating = self.dominating_level_transition_prob(
                    dominating_level, proposed_level)
                
                # Accept or reject the level move for both processes using the same uniform
                if u < accept_prob_dominating:
                    dominating_level = proposed_level
                    accepted_level_moves_dominating += 1
                    
                    # The actual process can only move if the dominating process moves
                    # AND its own acceptance criterion is met
                    if u < accept_prob_actual:
                        current_level = proposed_level
                        accepted_level_moves_actual += 1
                
            else:  # 50% chance to change parameter value
                # Propose a new parameter value
                proposed_theta_moves += 1
                proposed_theta = self.proposal_dist(current_theta)
                
                # Calculate acceptance probability
                accept_prob = self.within_level_transition_prob(
                    current_theta, proposed_theta, current_level)
                
                # Accept or reject the parameter move
                if np.random.random() < accept_prob:
                    current_theta = proposed_theta
                    accepted_theta_moves += 1
            
            # Store sample if past burn-in and at the right thinning interval
            if i >= burn_in and (i - burn_in) % thin == 0:
                # Store the levels for both processes
                dominating_levels.append(dominating_level)
                actual_levels.append(current_level)
                
                # Only store samples from the target distribution (level k where κ_k = 1)
                if current_level == self.num_levels - 1:
                    samples.append(current_theta)
        
        # Print diagnostics
        print(f"Actual level moves: {accepted_level_moves_actual}/{proposed_level_moves} "
              f"({accepted_level_moves_actual/proposed_level_moves:.2%} accepted)")
        print(f"Dominating level moves: {accepted_level_moves_dominating}/{proposed_level_moves} "
              f"({accepted_level_moves_dominating/proposed_level_moves:.2%} accepted)")
        print(f"Parameter moves: {accepted_theta_moves}/{proposed_theta_moves} "
              f"({accepted_theta_moves/proposed_theta_moves:.2%} accepted)")
        
        return np.array(samples), dominating_levels, actual_levels
    
    def perfect_sampling(self, max_iterations=10000):
        """
        Implement perfect sampling using the dominating process based on Algorithm A.58.
        
        The algorithm works by:
        1. Going backward far enough to find t* < 0 such that D_t* = N* and D_t' = 0 
           for at least one t* < t' <= 0
        2. Simulating the coupled chain (N_t, N_t) from t* steps with (N_t*, N_t*) = (0, 0)
        3. Taking (θ_0, N_0) as the final sample
        
        Parameters:
        -----------
        max_iterations : int, optional
            Maximum number of iterations to attempt for backward simulation
        
        Returns:
        --------
        tuple
            (sample, iterations) where:
            - sample is a perfect sample from the target distribution
            - iterations is the number of iterations required
        """
        # Step 1: Go backward far enough to find t* < 0
        # We'll simulate the dominating process backward in time
        # until we find a point where D_t* = N* (highest level)
        # and there's at least one point after it where D_t' = 0 (lowest level)
        
        t_star = None
        found_zero = False
        dominating_levels = []
        
        # Initialize at an arbitrary level
        current_level = np.random.randint(0, self.num_levels)
        dominating_levels.append(current_level)
        
        print("Starting backward simulation to find t*...")
        
        for t in range(1, max_iterations):
            # Propose a level move for the dominating process in reverse
            if current_level == 0:  # At the lowest level, can only go up
                proposed_level = 1
            elif current_level == self.num_levels - 1:  # At the highest level, can only go down
                proposed_level = self.num_levels - 2
            else:  # Can go either up or down
                # In reverse time, we swap the probabilities of up/down moves
                # to maintain detailed balance
                proposed_level = current_level + np.random.choice([-1, 1])
            
            # In reverse time, we swap the acceptance probabilities
            if proposed_level > current_level:  # This would be moving down in forward time
                accept_prob = self.dominating_level_transition_prob(
                    proposed_level, current_level)
            else:  # This would be moving up in forward time
                accept_prob = self.dominating_level_transition_prob(
                    proposed_level, current_level)
            
            # We use 1 - accept_prob for reverse moves to maintain detailed balance
            if np.random.random() < accept_prob:
                current_level = proposed_level
                
            dominating_levels.append(current_level)
            
            # Check if we've reached the highest level
            if current_level == self.num_levels - 1:
                # We've found a potential t*
                if found_zero:  # And we've also found a t' where D_t' = 0 after it
                    t_star = -t
                    break
            
            # Check if we've reached the lowest level
            if current_level == 0:
                found_zero = True
        
        if t_star is None:
            print(f"Failed to find suitable t* within {max_iterations} iterations.")
            return None, 0
        
        print(f"Found t* = {t_star}")
        
        # Step 2 & 3: Simulate the coupled chain (N_t, N_t) from t* steps with (N_t*, N_t*) = (0, 0)
        # and return the final state
        
        # We start both chains at level 0
        current_level = 0
        current_theta = 0.5  # Starting parameter value
        
        print("Simulating forward from t* with coupled chains...")
        
        for t in range(abs(t_star) + 1):
            # Simulate one step of the process
            # Decide whether to propose a level change or parameter change
            if np.random.random() < 0.5:  # 50% chance to change level
                # Propose a new level
                if current_level == 0:  # At the lowest level, can only go up
                    proposed_level = 1
                elif current_level == self.num_levels - 1:  # At the highest level, can only go down
                    proposed_level = self.num_levels - 2
                else:  # Can go either up or down
                    proposed_level = current_level + np.random.choice([-1, 1])
                
                # Calculate acceptance probability
                accept_prob = self.level_transition_prob(
                    current_theta, current_level, proposed_level)
                
                # Accept or reject the level move
                if np.random.random() < accept_prob:
                    current_level = proposed_level
            else:  # 50% chance to change parameter value
                # Propose a new parameter value
                proposed_theta = self.proposal_dist(current_theta)
                
                # Calculate acceptance probability
                accept_prob = self.within_level_transition_prob(
                    current_theta, proposed_theta, current_level)
                
                # Accept or reject the parameter move
                if np.random.random() < accept_prob:
                    current_theta = proposed_theta
        
        # After simulating forward, we should be at a perfect sample
        print("Perfect sampling completed.")
        
        return current_theta, abs(t_star)


# Example usage with a bimodal target distribution
def demo_power_tempering_with_domination():
    """
    Demonstrate the Power Tempering algorithm with dominating process on a bimodal distribution.
    """
    # Define a bimodal target distribution (mixture of two Gaussians)
    def bimodal_target(x):
        # Mixture of two Gaussians at -2 and 2 with standard deviation 0.5
        comp1 = stats.norm.pdf(x, loc=-2, scale=0.5)
        comp2 = stats.norm.pdf(x, loc=2, scale=0.5)
        return 0.5 * comp1 + 0.5 * comp2
    
    # Define a custom proposal distribution
    def proposal(x):
        return x + np.random.normal(0, 0.5)
    
    # Define a sequence of tempering powers
    kappas = [0.1, 0.3, 0.6, 1.0]  # κ_1 > ... > κ_4 = 1
    
    # Define coupling constants for the dominating process
    coupling_constants = [0.8, 0.9, 0.95]  # One fewer than the number of levels
    
    # Create the tempering sampler with dominating process
    sampler = PowerTemperingWithDomination(
        target_dist=bimodal_target,
        kappas=kappas,
        proposal_dist=proposal,
        coupling_constants=coupling_constants
    )
    
    print("Running standard MCMC with tempering and domination...")
    # Generate samples with the coupled dominating process
    samples, dom_levels, act_levels = sampler.sample_with_domination(
        n_samples=5000, burn_in=1000, thin=2)
    
    print("\nRunning perfect sampling algorithm...")
    # Generate a perfect sample
    perfect_sample, iterations = sampler.perfect_sampling(max_iterations=5000)
    
    # Plot the results
    plt.figure(figsize=(15, 10))
    
    # Plot the histogram of samples
    plt.subplot(2, 2, 1)
    plt.hist(samples, bins=50, density=True, alpha=0.7, label='MCMC Samples')
    
    # Plot the true target density
    x = np.linspace(-4, 4, 1000)
    true_density = [bimodal_target(xi) for xi in x]
    # Normalize the density (since our target is unnormalized)
    true_density = np.array(true_density) / np.trapz(true_density, x)
    plt.plot(x, true_density, 'r-', linewidth=2, label='True Density')
    
    # Mark the perfect sample if available
    if perfect_sample is not None:
        plt.axvline(perfect_sample, color='g', linestyle='--', linewidth=2, 
                   label=f'Perfect Sample: {perfect_sample:.3f}')
    
    plt.title('Power Tempering MCMC with Dominating Process')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    # Plot the tempered distributions
    plt.subplot(2, 2, 2)
    for i, kappa in enumerate(kappas):
        # Plot each tempered version of the distribution
        tempered_density = [bimodal_target(xi)**kappa for xi in x]
        # Normalize for visualization
        tempered_density = np.array(tempered_density) / np.trapz(tempered_density, x)
        plt.plot(x, tempered_density, label=f'κ_{i+1}={kappa}')
    
    plt.title('Tempered Distributions')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.legend()
    
    # Plot the level trajectories
    plt.subplot(2, 2, 3)
    # Only plot a subset of the iterations for clarity
    n_plot = min(1000, len(dom_levels))
    iterations = range(n_plot)
    plt.plot(iterations, dom_levels[:n_plot], 'b-', alpha=0.6, label='Dominating Process')
    plt.plot(iterations, act_levels[:n_plot], 'r-', alpha=0.6, label='Actual Process')
    
    plt.title('Level Trajectories')
    plt.xlabel('Iteration')
    plt.ylabel('Level')
    plt.legend()
    
    # Plot the distribution of levels
    plt.subplot(2, 2, 4)
    plt.hist(dom_levels, bins=np.arange(len(kappas)+1)-0.5, alpha=0.6, label='Dominating Process')
    plt.hist(act_levels, bins=np.arange(len(kappas)+1)-0.5, alpha=0.6, label='Actual Process')
    
    plt.title('Distribution of Levels')
    plt.xlabel('Level')
    plt.ylabel('Frequency')
    plt.xticks(range(len(kappas)))
    plt.legend()
    
    plt.tight_layout()
    plt.show()
