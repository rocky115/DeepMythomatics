## Metropolis-Hastings Algorithm and Irreducibility

The rationale behind this result is the following. If the conditional distribution $ q(y|x) $ allows for moves in a neighborhood of $ x^{(i)} $ with diameter bounded from below and if $ f $ is such that $ \rho(x^{(0)}, y) $ is positive in this neighborhood, then any subset of $ \mathcal{E} $ can be visited in $ k $ steps for $ k $ large enough. (This property obviously relies on the assumption that $ \mathcal{E} $ is connected.)

**Proof:**

Consider $ x^{(0)} $ an arbitrary starting point and $ A \subset \mathcal{E} $ an arbitrary measurable set. The connectedness of $ \mathcal{E} $ implies that there exist $ m \in \mathbb{N} $ and a sequence $ z^{(i)} \in \mathcal{E} $ $( 1 \leq i \leq m $) such that $ x^{(m)} \in A $ and $ |z^{(i+1)} - z^{(i)}| < \delta $. It is therefore possible to link $ x^{(0)} $ and $ A $ through a sequence of balls with radius $ \delta $. The assumptions on $ f $ imply that the acceptance probability of a point $ x^{(i)} $ of the $ i $-th ball starting from the $ (i-1) $-st ball is positive and, therefore,

$$
P_{x,m}^{(\alpha)}(A) = P(X^{(m)} \in A | X^{(0)} = x^{(0)}) > 0.
$$

By Theorem 6.15, the $ f $-irreducibility of $ (X^{(i)}) $ is established.

For an arbitrary value $ x^{(0)} \in \mathcal{E} $ and for every $ y \in B(x^{(0)}, \delta/2) $ (the ball with center $ x^{(0)} $ and radius $ \delta/2 $), we have

$$
P_{y}(A) \geq \int_{A} \rho(y, z) q(z | y) dz
$$

$$
= \int_{A \cap D_{y}} \frac{f(z)}{f(y)} q(y | z) dz + \int_{A \cap D_{y}} q(z | y) dz,
$$

where $ D_{y} = \{ z; f(z) q(y | z) \leq f(y) q(z | y) \} $. It therefore follows that

$$
P_{y}(A) \geq \int_{A \cap D_{y} \cap B} \frac{f(z)}{f(y)} q(y | z) dz + \int_{A \cap D_{y} \cap B} q(z | y) dz
$$

$$
\geq \frac{\inf_{B} f(z)}{\sup_{B} f(z)} \int_{A \cap D_{y} \cap B} q(y | z) dz + \int_{A \cap D_{y} \cap B} q(z | y) dz
$$

$$
\geq \varepsilon \frac{\inf_{B} f(z)}{\sup_{B} f(z)} \lambda(A \cap B),
$$

where $ \lambda $ denotes the Lebesgue measure on $ \mathcal{E} $. The balls $ B(x^{(0)}, \delta/2) $ are small sets associated with uniform distributions on $ B(x^{(0)}, \delta/2) $. This simultaneously implies the aperiodicity of $ (X^{(i)}) $ and the fact that every compact set is small.

$$

## Independent Metropolis-Hastings Algorithm

## Introduction

The Independent Metropolis-Hastings algorithm is a generalization of the Accept-Reject method. The instrumental distribution $ q $ is independent of $ X^{(i)} $ and is denoted $ g(y) $. The algorithm produces a transition from $ x^{(i)} $ to $ X^{(i+1)} $ as follows:

**Algorithm A.25** - Independent Metropolis-Hastings

Given $ x^{(i)} $:

1. Generate $ Y_i \sim g(y) $.
2. Take
   $$
   X^{(i+1)} = \begin{cases}
   Y_i & \text{with probability } \min\left\{\frac{f(Y_i)}{f(x^{(i)})} \frac{g(x^{(i)})}{g(Y_i)}, 1\right\} \\
   x^{(i)} & \text{otherwise}.
   \end{cases}
   $$

Although the $Y_i $'s are generated independently, the resulting sample is not iid. The probability of acceptance of $ Y_i $ depends on $ X^{(i)} $ (except in the trivial case when $ f = g $).

## Convergence Properties

The convergence properties of the chain $ (X^{(i)}) $ follow from properties of the density $ q $. The chain $ (X^{(i)}) $ is irreducible and aperiodic (thus, ergodic according to Corollary 7.5) if and only if $ q $ is almost everywhere positive on the support of $ f $.

Stronger properties of convergence, like geometric and uniform ergodicity, are described by the following result of Mengersen and Tweedie (1996):

**Theorem .8**: _The algorithm produces a uniformly ergodic chain if there exists a constant $ M $ such that_

$$
f(x) \leq M g(x), \quad \forall x \in \text{supp } f.
$$

_In this case, the chain converges uniformly to the target distribution._

## Convergence Bound in Total Variation Norm

The convergence of the Independent Metropolis-Hastings algorithm can be analyzed using the total variation norm. The following result provides a bound on the convergence rate:

**Theorem .8 (Continued)**:

If the condition $$ f(x) \leq M g(x) $$ holds for all $$ x \in \text{supp } f $$, then the total variation distance between the $$ n $$-step transition kernel $$ K^n(x, \cdot) $$ and the target distribution $$ f $$ satisfies:

$$
\|K^n(x, \cdot) - f\|_{TV} \leq 2 \left(1 - \frac{1}{M}\right)^n,
$$

where $$ \|\cdot\|_{TV} $$ denotes the total variation norm introduced in Definition 6.17.

### Proof

If the condition (7.7) is satisfied, the transition kernel $$ K(x, x') $$ satisfies:

$$
K(x, x') \geq g(x') \min\left\{\frac{f(x') g(x)}{f(x) g(x')}, 1\right\} = \min\left\{f(x') \frac{g(x)}{f(x)}, g(x')\right\} \geq \frac{1}{M} f(x').
$$

The set $$ \mathcal{E} $$ is therefore small, and the chain is uniformly ergodic (Theorem 6.59).

To establish the bound on $$ \|K^n(x, \cdot) - f\|_{TV} $$, we first write:

$$
\|K(x, \cdot) - f\|_{TV} = 2 \sup_{A} \left| \int_A (K(x, y) - f(y)) \, dy \right| = 2 \int_{\{y : f(y) \geq K(x, y)\}} (f(y) - K(x, y)) \, dy.
$$

This can be bounded as:

$$
\|K(x, \cdot) - f\|_{TV} \leq 2 \left(1 - \frac{1}{M}\right) \int_{\{y : f(y) \geq K(x, y)\}} f(y) \, dy \leq 2 \left(1 - \frac{1}{M}\right).
$$

We now use a recursion argument to establish the general bound (7.8). For $$ n = 2 $$, we have:

$$
\int_A (K^2(x, y) - f(y)) \, dy = \int_{\mathcal{E}} \left[ \int_A (K(u, y) - f(y)) \, dy \right] (K(x, u) - f(u)) \, du.
$$

Using a similar argument as above, we obtain:

$$
\|K^2(x, \cdot) - f\|_{TV} \leq 2 \left(1 - \frac{1}{M}\right)^2.
$$

The general recursion relation for $$ n + 1 $$ is:

$$
\int_A (K^{n+1}(x, y) - f(y)) \, dy = \int_{\mathcal{E}} \left[ \int_A (K^n(u, y) - f(y)) \, dy \right] (K(x, u) - f(u)) \, du.
$$

The proof of (7.8) is then established by induction.

## Convergence and Efficiency of Independent Metropolis-Hastings

### Conditions for Geometric Ergodicity

If the condition (7.7) does not hold, then the sets

$$
D_{n} = \left\{x, \frac{f(x)}{g(x)} \geq n\right\}
$$

satisfy $ P_{f}(D_{n}) > 0 $ for every $ n $. If $ x \in D_{n} $, then

$$
P(x, \{x\}) = 1 - \mathbb{E}_{q}\left[\min\left\{\frac{f(Y)g(x)}{g(Y)f(x)}, 1\right\}\right]
$$

$$
= 1 - P_{q}\left(\frac{f(Y)}{g(Y)} > \frac{f(x)}{g(x)}\right) - \mathbb{E}_{q}\left[\frac{f(Y)g(x)}{f(x)g(Y)}, \frac{1}{g(Y)} < \frac{f(x)}{g(x)}\right]
$$

$$
\geq 1 - P_{q}\left(\frac{f(Y)}{g(Y)} \geq n\right) - \frac{g(x)}{f(x)} \geq 1 - \frac{2}{n},
$$

since Markov's inequality implies that

$$
P_{q}\left(\frac{f(Y)}{g(Y)} \geq n\right) \leq \frac{1}{n} \mathbb{E}_{q}\left[\frac{f(Y)}{g(Y)}\right] = \frac{1}{n}.
$$

Consider a small set $ C $ such that $ D_{n} \cap C^{c} $ is not empty for $ n $ large enough and $ x_{0} \in D_{n} \cap C^{c} $. The return time to $ C $, $ \tau_{C} $, satisfies

$$
P_{x_{0}}(\tau_{C} > k) \geq \left(1 - \frac{2}{n}\right)^{k};
$$

therefore, the radius of convergence of the series (in $ s $) $ \mathbb{E}_{x_{0}}[s^{\tau_{C}}] $ is smaller than $ n/(n-2) $ for every $ n $, and this implies that $ (X^{(t)}) $ cannot be geometrically ergodic, according to Theorem 6.75.

### Expected Acceptance Probability

**Lemma 7.9**: _If (7.7) holds, the expected acceptance probability associated with the algorithm [4, 25] is at least $ \frac{1}{M} $ when the chain is stationary._

Proof: If the distribution of $ f(X)/g(X) $ is absolutely continuous, the expected acceptance probability is

$$
\mathbb{E}\left[\min\left\{\frac{f(Y_{i})g(X^{(t)})}{f(X^{(t)})g(Y_{i})}, 1\right\}\right] = \int \mathbb{I}_{\{\frac{f(y)g(x)}{g(y)f(x)} > 1\}} f(x)g(y) \, dx \, dy
$$

$$
+ \int \frac{f(y)g(x)}{g(y)f(x)} \mathbb{I}_{\{\frac{f(y)g(x)}{g(y)f(x)} \leq 1\}} f(x)g(y) \, dx \, dy
$$

$$
= 2 \int \mathbb{I}_{\{\frac{f(y)g(x)}{g(y)f(x)} \geq 1\}} f(x)g(y) \, dx \, dy
$$

$$
\geq 2 \int \mathbb{I}_{\{\frac{f(y)g(x)}{g(y)f(x)} \geq 1\}} f(x) \frac{f(y)}{M} \, dx \, dy
$$

$$
= \frac{2}{M} P\left(\frac{f(X_{1})}{g(X_{1})} \geq \frac{f(X_{2})}{g(X_{2})}\right) = \frac{1}{M}.
$$

Since $ X_{1} $ and $ X_{2} $ are independent and distributed according to $ f $, this last probability is equal to $ 1/2 $, and the result follows.

### Comparison with Accept-Reject Methods

The independent Metropolis-Hastings algorithm [4, 25] is more efficient than the Accept-Reject algorithm [4, 4] in its handling of the sample produced by $ g $, since, on average, it accepts more proposed values. A more advanced comparison between these approaches is about as difficult as the comparison between Accept-Reject and importance sampling proposed in Section 3.3.3, namely that the size of one of the two samples is random, and this complicates the computation of the variance of the resulting estimator. In addition, the correlation between the $ X_{1} $'s resulting from [4, 25] prohibits a closed-form expression of the joint distribution. We therefore study the consequence of the correlation on the variance of both estimators through an example.

**Example 7.10**: **Generating gamma variables.** Using the algorithm of Example 2.19 (see also Example 3.15), an Accept-Reject method can be derived to generate random variables from the $ \mathcal{G}a(a, \beta) $ distribution using a Gamma $ \mathcal{G}a(|a|, b) $ candidate (where $ |a| $ denotes the integer part of $ a $). When $ \beta = 1 $, the optimal choice of $ b $ is

````python
import numpy as np
import matplotlib.pyplot as plt

def generate_gamma(a, beta, b):
    while True:
        u = np.random.uniform(0, 1)
        v = np.random.uniform(0, 1)
        x = b * np.log(1 / v)
        if u <= (x ** (a - 1)) * np.exp(-x):
            return x

def accept_reject_gamma(a, beta, b, num_samples):
    samples = [generate_gamma(a, beta, b) for _ in range(num_samples)]
    return samples

def independent_metropolis_hastings_gamma(a, beta, b, num_samples):
    samples = [b * np.log(1 / np.random.uniform(0, 1)) for _ in range(num_samples)]
    return samples

# Example parameters
a = 2
beta = 1
b = 1
num_samples = 10000

# Accept-Reject method
accept_reject_samples = accept_reject_gamma(a, beta, b, num_samples)

# Independent Metropolis-Hastings method
metropolis_hastings_samples = independent_metropolis_hastings_gamma(a, beta, b, num_samples)

# Plot the histograms of the samples
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(accept_reject_samples, bins=50, density=True, label="Accept-Reject")
plt.title("Accept-Reject")
plt.xlabel("Vabylue")
plt.ylabel("Probability Density")

plt.subplot(1, 2, 2)
plt.hist(metropolis_hastings_samples, bins=50, density=True, label="Independent Metropolis-Hastings")
plt.title("Independent Metropolis-Hastings")
plt.xlabel("Value")
plt.ylabel("Probability Density")

plt.tight_layout()
plt.show()
````
