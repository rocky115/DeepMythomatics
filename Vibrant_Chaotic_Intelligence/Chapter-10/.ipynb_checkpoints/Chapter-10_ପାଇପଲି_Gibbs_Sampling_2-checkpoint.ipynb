## Convergence Properties of Gibbs Samplers

This section discusses the convergence properties of Gibbs samplers, particularly focusing on Harris recurrence and convergence in total variation.

### Harris Recurrence

**Lemma 10.9:** If the transition kernel associated with [A.40] is absolutely continuous with respect to the dominating measure, the resulting chain is Harris recurrent.

*Proof:*
The proof is similar to that of Lemma 7.3, where Harris recurrence was shown to follow from irreducibility. Here, the condition on the Gibbs transition kernel yields an irreducible chain, and Harris recurrence follows. Once the irreducibility of the chain associated with [4.40] is established, more advanced convergence properties can be established.

**Remark:**
This condition of absolute continuity on the kernel (10.2) is satisfied by most decompositions. However, if one of the $i$ steps $(1 \le i \le p)$ is replaced by a simulation from a Metropolis-Hastings algorithm, as in the hybrid algorithms described in Section 10.3, absolute continuity is lost and it will be necessary to either study the recursion properties of the chain or to introduce an additional simulation step to guarantee Harris recurrence.

### Convergence in Total Variation

From the property of Harris recurrence, we can now establish a result similar to Theorem 7.4.

**Theorem 10.10:** If the transition kernel of the Gibbs chain $(Y^{(t)})$ is absolutely continuous with respect to the measure $\mu$, then:

**(i)** If $h_1, h_2 \in L^1(g)$ with $\int h_2(y) dg(y) \ne 0$, then
$$
\lim_{n \to \infty} \frac{\sum_{t=1}^T h_1(Y^{(t)})}{\sum_{t=1}^T h_2(Y^{(t)})} = \frac{\int h_1(y) dg(y)}{\int h_2(y) dg(y)} \quad \text{a.e. } g.
$$

**(ii)** If, in addition, $(Y^{(i)})$ is aperiodic, then, for every initial distribution $\mu$,
$$
\lim_{n \to \infty} ||\int K^n(y, \cdot) \mu(dx) - g||_{TV} = 0.
$$




**Explanation of the Results:**

* **Harris recurrence** essentially ensures that the chain will visit every part of the state space (in a suitable sense) infinitely often. This is a crucial property for establishing convergence to the stationary distribution.

* **Part (i) of Theorem 10.10** shows that ratios of time averages of functions converge to the ratios of their expectations under the stationary distribution. This is a type of ergodic theorem.

* **Part (ii) of Theorem 10.10** establishes convergence in total variation (TV) of the distribution of the chain to the stationary distribution. This means that the probability measures of the chain's state get arbitrarily close to the stationary distribution as time goes on, uniformly over all possible sets.  The aperiodicity condition is needed to ensure that the chain doesn't get "stuck" in periodic cycles.

In simpler terms, these results tell us that under suitable conditions, the Gibbs sampler will eventually explore the entire distribution and that the distribution of the samples it generates will converge to the target distribution as the number of samples increases.

```python
import random
import matplotlib.pyplot as plt

# Illustrative implementation (simplified) of Gibbs sampling and convergence

def gibbs_sampler_2d_gaussian(mean, cov, num_samples):
    """
    Illustrative Gibbs sampler for a 2D Gaussian.  
    IMPORTANT: This is simplified. True Gibbs for a Gaussian uses conditional 
    distributions derived from the joint distribution.
    """
    samples = []
    x, y = mean  # Initialize at the mean (for simplicity)

    for _ in range(num_samples):
        # Simplified conditionals (replace with actual conditionals for real use)
        x = random.gauss(mean[0], cov[0][0]**0.5)  # Sample x given y (simplified)
        y = random.gauss(mean[1], cov[1][1]**0.5)  # Sample y given x (simplified)
        samples.append([x, y])
    return samples

def plot_samples_and_convergence(samples, true_mean, true_cov):
    """Plots the samples and illustrates convergence."""
    x_vals = [s[0] for s in samples]
    y_vals = [s[1] for s in samples]

    plt.figure(figsize=(8, 6))
    plt.scatter(x_vals, y_vals, s=10, alpha=0.5, label="Gibbs Samples")

    # Plot the true distribution (for comparison)
    # (Requires NumPy for multivariate_normal, but can be approximated without)
    try:
      import numpy as np
      from scipy.stats import multivariate_normal
      x, y = np.mgrid[mean[0]-3*cov[0][0]**0.5:mean[0]+3*cov[0][0]**0.5:.01, mean[1]-3*cov[1][1]**0.5:mean[1]+3*cov[1][1]**0.5:.01]
      pos = np.dstack((x, y))
      rv = multivariate_normal(mean, cov)
      plt.contourf(x, y, rv.pdf(pos), levels=10, cmap='viridis', alpha=0.3, label="True Distribution")
    except ImportError:
      print("NumPy and SciPy not found. Cannot plot true distribution contour.")


    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Gibbs Sampler Convergence (Illustrative)")
    plt.legend()
    plt.show()


    # Illustrate convergence (simplified - showing running mean)
    running_means_x = [sum(x_vals[:i+1]) / (i+1) for i in range(len(x_vals))]
    running_means_y = [sum(y_vals[:i+1]) / (i+1) for i in range(len(y_vals))]

    plt.figure(figsize=(8, 6))
    plt.plot(running_means_x, label="Running Mean X")
    plt.plot(running_means_y, label="Running Mean Y")
    plt.axhline(y=true_mean[0], color='r', linestyle='--', label="True Mean X")
    plt.axhline(y=true_mean[1], color='g', linestyle='--', label="True Mean Y")

    plt.xlabel("Iteration")
    plt.ylabel("Running Mean")
    plt.title("Convergence of Running Means")
    plt.legend()
    plt.show()



# Example Usage:
mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]  # Example covariance matrix
num_samples = 2000

samples = gibbs_sampler_2d_gaussian(mean, cov, num_samples)
plot_samples_and_convergence(samples, mean, cov)


# Example with a different covariance (to show how it affects convergence)
cov2 = [[2, 0.9], [0.9, 2]]
samples2 = gibbs_sampler_2d_gaussian(mean, cov2, num_samples)
plot_samples_and_convergence(samples2, mean, cov2)
```

## Gibbs Sampling and Metropolis-Hastings Algorithm

The provided text discusses Gibbs sampling and its relation to the Metropolis-Hastings algorithm. Here's a breakdown in a Jupyter Notebook friendly Markdown format, incorporating LaTeX for mathematical expressions:

---

### Gibbs Sampling Property

Gibbs sampling is a Markov Chain Monte Carlo (MCMC) method that iteratively samples from the conditional distributions of a joint probability distribution. A key property is that the acceptance probability is always 1. 

Mathematically, if we denote the joint distribution as $g(y) = g(y_1, y_2, ..., y_p)$, and the conditional distributions as $g_i(y_i | y_{-i})$ (where $y_{-i}$ represents all variables except $y_i$), the transition kernel can be expressed as:

$$
\frac{g(y')}{g(y)} \frac{q(y, y')}{q(y', y)} = 1 
$$

where $q(y, y')$ is the proposal distribution.

This simplifies the acceptance probability calculation, making Gibbs sampling efficient.

---

### Metropolis-Hastings and Acceptance Probability

While Gibbs sampling inherently has an acceptance probability of 1, this is not the case for all MCMC algorithms. The text mentions that any kernel $q$ associated with a reversible Markov chain with invariant distribution $g$ has an acceptance probability of 1. However, when considering a vector of parameters, the global acceptance probability for a particular update (like in the Metropolis-Hastings algorithm) is usually different from 1.

The text provides an example where a direct processing as a Metropolis-Hastings algorithm leads to a positive probability of rejection.

---

### Example: Two-Dimensional Autoexponential Model

Consider a two-dimensional autoexponential model with joint distribution:

$$
g(y_1, y_2) \propto \exp\{-y_1 - y_2 - \theta_{12} y_1 y_2\}
$$

The conditional densities are:

$$
g_1(y_1 | y_2) = (1 + \theta_{12}y_2) \exp\{-(1 + \theta_{12}y_2)y_1\}
$$

$$
g_2(y_2 | y_1) = (1 + \theta_{12}y_1) \exp\{-(1 + \theta_{12}y_1)y_2\}
$$

The ratio used in the Metropolis-Hastings algorithm is:

$$
\frac{g(y_1', y_2') K((y_1, y_2), (y_1', y_2'))}{g(y_1, y_2) K((y_1', y_2'), (y_1, y_2))} = \frac{(1 + \theta_{12}y_2')(1 + \theta_{12}y_1)}{(1 + \theta_{12}y_2)(1 + \theta_{12}y_1')} e^{\theta_{12}(y_1'y_2 - y_1y_2')}
$$

This ratio is generally not equal to 1, indicating that the acceptance probability in a Metropolis-Hastings update will not always be 1.

---

### Comparison and Further Research

The text suggests that directly applying Metropolis-Hastings to the conditional updates (as opposed to the inherent Gibbs sampling approach) introduces rejections. This version could be compared to the original Gibbs sampler. However, the text mentions that no full-scale study has been undertaken in this direction, except for modifications introduced by Liu (1995, 1996a).

The fact that the Metropolis-Hastings approach allows for rejections might be beneficial in some scenarios, but the standard Gibbs sampling algorithm doesn't have this characteristic.

## Lemma .11

Let $ y = (y_1, \dots, y_p) $ and $ y' = (y'_1, \dots, y'_p) $ and suppose there exists $ \delta > 0 $ for which $ y, y' \in \text{supp}(g) $, $ |y - y'| \leq \delta $, and  

$$
g(y_i|y_1, \dots, y_{i-1}, y'_{i+1}, \dots, y'_p) > 0, \quad i = 1, \dots, p.

If there exists $ \delta < 5 $ for which almost every pair $ (y, y') \in \text{supp}(g) $ can be connected by a finite sequence of balls with radius $ \delta $, having the (stationary) measure of the intersections of two consecutive balls positive, then the chain produced by $ A[4.40] $ is **irreducible and aperiodic**.

The laborious formulation of this condition is needed to accommodate settings where the support of $ g $ is not connected, as in Example .7, since a necessary irreducibility condition is that **two connected components** of $ \text{supp}(g) $ can be linked by the kernel of $ A[4.40] $. ## Gibbs Sampling as Metropolis–Hastings

We now examine the **exact relationship** between the Gibbs sampling and Metropolis-Hastings algorithms by considering $ A[4.40] $ as the composition of $ p $ Markovian kernels. (As mentioned earlier, this representation is not sufficient to establish irreducibility since each of these separate Metropolis-Hastings algorithms does not produce an irreducible chain, given that it only modifies one component of $ y $. In fact, these kernels are **never irreducible**, since they are constrained in subspaces of lower dimensions.)

---

## Theorem .13

The **Gibbs sampling method** of $ A[4.40] $ is equivalent to the composition of $ p $ **Metropolis-Hastings algorithms**, with **acceptance probabilities uniformly equal to 1**.

#### **Proof**
If we write $ A[4.40] $ as the composition of $ p $ "elementary" algorithms that correspond to the $ p $ simulation steps from the conditional distribution, it is sufficient to show that each of these algorithms has an **acceptance probability of 1**.

For $ 1 \leq i \leq p $, the instrumental distribution in step $ i $ of $ A[4.40] $ is given by:

$$
q_i(y_i' | y) = \delta_{y_1, \dots, y_{i-1}, y_{i+1}, \dots, y_p}(y'_1, \dots, y'_{i-1}, y'_{i+1}, \dots, y'_p)
$$

$$
\times g_i(y'_i | y_1, \dots, y_{i-1}, y_{i+1}, \dots, y_p)
$$

And the ratio defining the probability \( p(y, y') \) is therefore:

$$
p(y, y') = 1.
$$


```python
import random

class GibbsSamplerMH:
    def __init__(self, joint_distribution, initial_state, num_samples=1000):
        """
        Gibbs sampler implementation using the Metropolis-Hastings framework.

        :param joint_distribution: Function that returns conditional probabilities.
        :param initial_state: Initial values for the variables.
        :param num_samples: Number of samples to generate.
        """
        self.joint_distribution = joint_distribution
        self.state = initial_state[:]
        self.num_samples = num_samples
        self.num_vars = len(initial_state)

    def sample(self):
        """ Runs the Gibbs sampler based on Metropolis-Hastings principles. """
        samples = []

        for _ in range(self.num_samples):
            for i in range(self.num_vars):
                self.state[i] = self._sample_from_conditional(i)
            samples.append(self.state[:])  # Store the sample (copy)

        return samples

    def _sample_from_conditional(self, i):
        """
        Sample a new value for state[i] using conditional distribution.

        :param i: Index of the variable to sample.
        :return: A new sampled value for state[i].
        """
        y_copy = self.state[:]
        possible_values = [0, 1]  # Example: binary case

        probabilities = [self.joint_distribution(i, val, y_copy) for val in possible_values]
        total_prob = sum(probabilities)

        if total_prob == 0:
            return random.choice(possible_values)  # Avoid division by zero

        probabilities = [p / total_prob for p in probabilities]  # Normalize

        return random.choices(possible_values, probabilities)[0]

# Example joint distribution function (binary case)
def example_joint_distribution(i, y_i, y):
    """
    Example joint probability distribution function.
    
    :param i: Index of the variable to condition on.
    :param y_i: Value to sample.
    :param y: Current state.
    :return: Probability of y_i given other variables.
    """
    y_temp = y[:]
    y_temp[i] = y_i
    return 1.0 / (1 + sum(y_temp))  # Example probability function

# Example usage
initial_state = [1, 0, 1, 0]
sampler = GibbsSamplerMH(example_joint_distribution, initial_state, num_samples=10)
samples = sampler.sample()

print("Generated Samples:")
for s in samples:
    print(s)
```





##  Hierarchical Structures

To conclude this section, we investigate a structure for which Gibbs sampling is particularly well adapted, that of **hierarchical models**. These are structures in which the distribution $ f $ can be **decomposed** as $( \ell \geq 1 $):

$$
f(x) = \int f_1(x_1|z_1) f_2(x_2|z_2) \cdots f_\ell(x_\ell+1|z_\ell) dz_1 \cdots dz_{\ell+1}.
$$

Such models naturally appear in **Bayesian analysis** of complex models, where the diversity of prior information or the variability of the observations may require the introduction of several levels of **prior distributions** (see Wakefield et al. 1994, Robert 2001, Bennett et al. 1996, Spiegelhalter et al. 1996, Draper 1998).  

This is, for instance, the case of **Example .17**, where the first level represents the **exchangeability** of the parameters $ \lambda_i $, and the second level represents the **prior information**.  

In particular, where the **prior information** is sparse, **hierarchical modeling** is also useful since diffuse (or **noninformative**) distributions can be introduced at various levels of the hierarchy.

The following two examples illustrate situations where hierarchical models are particularly useful (see also **Problems 10.29–10.36**).

---

## Example 10.15: Animal Epidemiology

Research in **animal epidemiology** sometimes uses data from groups of animals, such as **litters** or **herds**. Such data **may not** follow the usual assumptions of **independence**, etc., and as a result, **variances of parameter estimates tend to be larger** (this phenomenon is also referred to as **"overdispersion"**).  

### **Schukken et al. (1991) obtained counts** of the number of cases of **clinical mastitis** in dairy cattle herds over a year period.

If we assume that, in each **herd**, the occurrence of mastitis is a **Bernoulli random variable**, and if we let $ X_{i} $, $ i = 1, \dots, m $, **denote the number of cases** in herd $ i $, it is then reasonable to model:

$$
X_i \sim \mathcal{P}(\lambda_i),
$$

where $ \lambda_i $ is the **underlying rate of infection** in herd $ i $. However, **there is lack of independence here** (mastitis is infectious), which **might manifest itself as overdispersion**.  

To account for this, **Schukken et al. (1991) put a gamma prior distribution** on the **Poisson parameter**.

A **complete hierarchical specification** is:

$$
X_i \sim \mathcal{P}(\lambda_i),
$$

$$
\lambda_i \sim \text{Ga}(\alpha, \beta),
$$

$$
\beta_i \sim \text{Ga}(a, b),
$$

where $ a, b, \alpha, $ and $ u $ are user-specified.  

The **posterior density** of $ \lambda_i $, given $ X_i $, can now be simulated **via the Gibbs sampler**:

$$
\lambda_i \sim \mathcal{P}(\lambda_i | X_i) = \mathcal{G}(\alpha + n, 1 + \beta_i),
$$

$$
\beta_i \sim \mathcal{P}(\beta_i | X_i, a, b, \lambda_i) = \mathcal{G}(a + \alpha, b + \lambda_i).
$$

## Example .16: Medical Models

Part of the concern of the study of **pharmacokinetics** is the modeling of the **relationship between the dosage of a drug and the resulting concentration in the blood**. (More generally, **pharmacokinetics studies the different interactions of a drug and the body**.)

**Gilks et al. (1993)** introduce an approach for estimating **pharmacokinetic parameters** that **uses the traditional mixed-effects model and nonlinear structure**, but which is also **robust to outliers** common to **clinical trials**.

For a given **dose** $ d_i $ administered at time $ t_0 $ to patient $ i $, the **measured log concentration** in the blood at time $ t_{ij} $, $ X_{ij} $, is assumed to follow a **Student's t distribution**:

$$
X_{ij} - \log g(y_i(\lambda_i)) \sim T_n,
$$

where:

- $ g(y_i(\lambda_i)) $ represents the **drug concentration model**.
- $ T_n $ is a **Student's t-distribution**.



```python
import random
import math

# Gamma Sampling using Inverse Transform
def gamma_sample(shape, scale):
    """ Sample from a Gamma distribution using the Marsaglia-Tsang method. """
    if shape < 1:  # Use transformation for shape < 1
        shape += 1
        u = random.uniform(0, 1)
        return gamma_sample(shape, scale) * u ** (1 / shape)

    d = shape - 1 / 3
    c = 1 / math.sqrt(9 * d)
    while True:
        x = random.gauss(0, 1)
        v = (1 + c * x) ** 3
        u = random.uniform(0, 1)
        if x > -1 / c and math.log(u) < 0.5 * x ** 2 + d * (1 - v + math.log(v)):
            return d * v / scale

# Poisson Sampling using Inverse Transform
def poisson_sample(lmbda):
    """ Sample from a Poisson distribution using the Knuth algorithm. """
    L = math.exp(-lmbda)
    k, p = 0, 1
    while p > L:
        k += 1
        p *= random.uniform(0, 1)
    return k - 1

# Gibbs Sampler
def gibbs_sampler(n_iters, alpha, beta_init, a, b, X):
    """ Gibbs sampler for the hierarchical Poisson-Gamma model. """
    m = len(X)
    lambda_samples = [1.0] * m  # Initialize lambda values
    beta = beta_init
    samples = []

    for _ in range(n_iters):
        # Sample lambda_i given X_i and beta
        for i in range(m):
            lambda_samples[i] = gamma_sample(alpha + X[i], beta + 1)

        # Sample beta given lambda_i
        beta = gamma_sample(a + alpha * m, b + sum(lambda_samples))

        samples.append((lambda_samples[:], beta))  # Store samples

    return samples

# Example Data
X = [3, 5, 2, 7, 4]  # Number of mastitis cases in herds
alpha, beta_init, a, b = 2.0, 1.0, 1.0, 1.0  # Hyperparameters
n_iters = 5000

# Run Gibbs Sampler
samples = gibbs_sampler(n_iters, alpha, beta_init, a, b, X)

# Print Final Estimates
print("Final Lambda Estimates:", [round(l, 3) for l in samples[-1][0]])
print("Final Beta Estimate:", round(samples[-1][1], 3))
```

## Hierarchical Models and Gibbs Sampling: An Example

This section delves into hierarchical models and the application of Gibbs sampling, focusing on a specific example involving clearance and volume parameters.

---

## Model Setup

Consider a hierarchical model where:

-  $\lambda_{i} = (\log C_i, \log V_i)'$ represents a vector of parameters for the i-th individual, with $C_i$ denoting clearance and $V_i$ denoting volume.
-  $\sigma^2$ is the measurement error variance.
-  $g_{ij}(\lambda_i) = \frac{d_i}{V_i} \exp(-\frac{C_i t_{ij}}{V_i})$

The hierarchy is completed by specifying priors:

- $\pi(\sigma) = 1/\sigma$ (non-informative prior on $\sigma$)
- $\lambda_i \sim \mathcal{N}(\theta, \Sigma)$
- $\theta \sim \mathcal{N}(\tau_1, T_1)$
- $\Sigma^{-1} \sim \mathcal{W}_2(\tau_2, T_2)$

where $\tau_1, T_1, \tau_2$, and $T_2$ are specified hyperparameters.

---

## Conjugate Structures and Student's t-distribution

To facilitate Gibbs sampling, conjugate structures are exploited.  Using Dickey's decomposition of the Student's t-distribution, we introduce artificial variables $\omega_{ij}$ such that:

$$X_{ij} | \omega_{ij} \sim \mathcal{N}(\log g_{ij}(\lambda_i), \omega_{ij} \sigma^2 \frac{n}{n-2})$$

This leads to:

- Full conditional distributions for $C_i$ and $\theta$ that are normal.
- Full conditional distributions for $\sigma^2$ and $\Sigma$ that are inverse gamma and inverse Wishart, respectively.

---

## Challenges with $V_i$ and Solutions

The full conditional distribution for $V_i$ is more complex:

$$
\propto \exp \left\{ -\frac{1}{2} \sum_j \left( \frac{(\log V_i + \frac{C_i t_{ij}}{V_i} - \mu_{iv})^2}{\xi} + \frac{(V_i - \gamma_i)^2}{\zeta} \right) \right\}  \quad (10.4)
$$

where $\mu_{iv}, \xi, \gamma_i$, and $\zeta$ are hyperparameters dependent on other parameters and $x_{ij}$.

**Solutions:**

1. **Accept-Reject Algorithm:**  As suggested by Gilks et al. (1993), remove the $\frac{C_i t_{ij}}{V_i}$ terms for simplification.

2. **Metropolis-Hastings Step:**  Apply a Metropolis-Hastings step as described in Section 10.3.3.

---

## Ergodicity and Convergence

For some hierarchical models, uniform ergodicity of the Gibbs chains can be shown, often requiring case-by-case analysis. Schervish and Carlin (1992) studied geometric convergence, which involves conditions on the kernel that are challenging to assess.
