## Reversible and Interleaving Chains

This section delves into the properties of two-stage Gibbs samplers, particularly the concept of **interleaving**.

**Interleaving Property (Definition 9.10):**

Two Markov chains \((X^{(t)})\) and \((Y^{(t)})\) are **interleaved** (or conjugate) if they satisfy the following conditions:

1.  $X^{(t)}$ and $X^{(t+1)}$ are conditionally independent given $Y^{(t)}$.
2.  $Y^{(t-1)}$ and $Y^{(t)}$ are conditionally independent given $X^{(t)}$.
3.  The joint distributions $(X^{(t)}, Y^{(t-1)})$ and $(X^{(t+1)}, Y^{(t)})$ are identical under stationarity.

In essence, interleaving implies that each chain updates independently when the other chain's state is known, and there's a specific symmetry in the joint distribution across time steps.

**Interleaving in Two-Stage Gibbs Samplers:**

The text states that the interleaving property is inherent in two-stage Gibbs samplers. This is because the updates in each stage are conditionally independent given the variables from the other stage.

**Reversibility (Lemma .11):**

Each individual chain $(X^{(t)})$ and $(Y^{(t)})$ generated by a two-stage Gibbs sampler is **reversible**. This follows from the detailed balance condition (Theorem 6.46).

**Joint Chain Reversibility:**

However, the joint chain $(X^{(t)}, Y^{(t)})$ is *not* necessarily time-reversible, even though the individual chains are.

**Significance of Interleaving:**

The interleaving property is a crucial structural characteristic of two-stage Gibbs samplers. It highlights the specific way the two "sub-chains" interact and update, and it has implications for the analysis of convergence and other properties of the sampler.

**Transition Kernel:**

The transition kernel $K_1(x_0, x_1)$ for the chain $(X^{(t)})$ is defined as:

$$
K_1(x_0, x_1) = \int g_2(y_0 | x_0) g_1(x_1 | y_0) dy_0
$$

where $g_1$ and $g_2$ are the conditional distributions in the two stages of the Gibbs sampler.  This kernel describes the probability of transitioning from state $x_0$ to $x_1$ in one step of the chain.

## Algorithm A.36: Reversible Two-Stage Gibbs Sampler

Given $y_2^{(t)}$,

1.  Simulate $W \sim g_1(w | y_2^{(t)})$
2.  Simulate $Y_2^{(t+1)} \sim g_2(y_2^{(t+1)} | w)$ 
3.  Simulate $Y_1^{(t+1)} \sim g_1(y_1^{(t+1)} | y_2^{(t+1)})$

In this case,

$$
(Y_1, Y_2, Y_1', Y_2') \sim g(y_1, y_2) \int g_1(w | y_2) g_2(y_2' | w) dw \, g_1(y_1' | y_2')
$$

$$
= g(y_1', y_2') \int \frac{g_2(y_2' | w)}{g^{(1)}(y_2') } g(w, y_2) dw \, g_1(y_1 | y_2)
$$

$$
= g(y_1', y_2') \int g_1(w | y_2') g_2(y_2 | w) dw \, g_1(y_1 | y_2)
$$

is distributed as $(Y_1', Y_2', Y_1, Y_2)$.

```python
import random
import numpy as np  # NumPy is used for array manipulation (optional, but recommended)
from scipy.stats import norm  # SciPy for normal distribution (example)

# Example: Reversible Two-Stage Gibbs Sampler for a toy problem

def reversible_two_stage_gibbs(n_iter=10000):
    """
    Implements a reversible two-stage Gibbs sampler.

    This is a simplified example.  Adapt the distributions and logic
    to your specific problem.  This example uses normal distributions.
    """

    y1_current = 0.0  # Initial values
    y2_current = 0.0

    y1_samples = []
    y2_samples = []

    for _ in range(n_iter):
        # 1. Simulate W (intermediate variable)
        # Example: W | Y2 ~ N(Y2, 1)  (Normal distribution)
        w = norm.rvs(loc=y2_current, scale=1)  # Using scipy.stats.norm

        # 2. Simulate Y2_next
        # Example: Y2_next | W ~ N(W, 1)
        y2_next = norm.rvs(loc=w, scale=1)

        # 3. Simulate Y1_next
        # Example: Y1_next | Y2_next ~ N(Y2_next, 1)
        y1_next = norm.rvs(loc=y2_next, scale=1)

        y1_samples.append(y1_next)
        y2_samples.append(y2_next)

        # Update current values for the next iteration
        y1_current = y1_next
        y2_current = y2_next

    return y1_samples, y2_samples


y1_samples, y2_samples = reversible_two_stage_gibbs(n_iter=10000)

# Analyze Results (Example)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(y1_samples, alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("Y1")
plt.title("Y1 Samples")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(y2_samples, alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("Y2")
plt.title("Y2 Samples")
plt.grid(True)

plt.tight_layout()
plt.show()


# Joint Distribution
plt.figure(figsize=(8, 6))
plt.scatter(y1_samples, y2_samples, alpha=0.3, s=10)
plt.xlabel("Y1")
plt.ylabel("Y2")
plt.title("Joint Distribution of Y1 and Y2")
plt.grid(True)
plt.show()
```

##  The Duality Principle

The well-known concept of Rao-Blackwellization (Sections 4.2, 7.6.2, and to come in Section 9.3) is based on the fact that a conditioning argument, using variables other than those directly of interest, can result in an improved procedure. In the case of interleaving Markov chains, this phenomenon goes deeper. We will see in this section that this phenomenon is more fundamental than a mere improvement of the variance of some estimators, as it provides a general technique to establish convergence properties for the chain of interest $(X^{(t)})$ based on the instrumental chain $(Y^{(t)})$, even when the latter is unrelated with the inferential problem. Diebolt and Robert (1993, 1994) have called this use of the dual chain the *Duality Principle* when they used it in the setup of mixtures of distributions. While it has been introduced in a two-stage Gibbs sampling setup, this principle extends to other Gibbs sampling methods since it is sufficient that the chain of interest $(X^{(t)})$ be generated conditionally on another chain $(Y^{(t)})$, which supplies the probabilistic properties.

**Theorem 9.12.** Consider a Markov chain $(X^{(t)})$ and a sequence $(Y^{(t)})$ of random variables generated from the conditional distributions

$$
\begin{aligned}
X^{(t)} | Y^{(t)} &\sim \pi(x | y^{(t)}), \\
Y^{(t+1)} | X^{(t)}, Y^{(t)} &\sim f(y^{(t+1)} | x^{(t)}, y^{(t)}).
\end{aligned}
$$

If the chain $(Y^{(t)})$ is ergodic (geometrically or uniformly ergodic) and if $\pi_m^{(0)}$ denotes the distribution of $(X^{(t)})$ associated with the initial value $y^{(0)}$, the norm $|| \pi_m^{(t)} - \pi ||_{TV}$ goes to 0 when $t$ goes to infinity (goes to 0 at a geometric or uniformly bounded rate).

**Proof.** The transition kernel is
$$
K(y, y') = \int_X \pi(x|y) f(y'|x, y) dx.
$$

If $f_{y^{(0)}}^t(y)$ denotes the marginal density of $Y^{(t)}$, then the marginal of $X^{(t)}$ is
$$
\pi_{y^{(0)}}^t(x) = \int_{\mathcal{Y}} \pi(x|y) f_{y^{(0)}}^t(y) dy
$$
and convergence in the $(X^{(t)})$ chain can be tied to convergence in the $(Y^{(t)})$ chain by the following. The total variation in the $(X^{(t)})$ chain is
$$
\begin{aligned}
|| \pi_{y^{(0)}}^t - \pi ||_{TV} &= \frac{1}{2} \int_X | \pi_{y^{(0)}}^t(x) - \pi(x) | dx \\
&= \frac{1}{2} \int_X \left| \int_{\mathcal{Y}} \pi(x|y) (f_{y^{(0)}}^t(y) - f(y)) dy \right| dx \\
&\le \frac{1}{2} \int_X \int_{\mathcal{Y}} | f_{y^{(0)}}^t(y) - f(y) | \pi(x|y) dx dy \\
&= || f_{y^{(0)}}^t - f ||_{TV}.
\end{aligned}
$$

So the convergence properties of $|| f_{y^{(0)}}^t - f ||_{TV}$ can be transferred to $|| \pi_{y^{(0)}}^t - \pi ||_{TV}$. Both sequences have the same speed of convergence since
$$
|| f_{y^{(0)}}^t - f ||_{TV} \le || \pi_{y^{(0)}}^t - \pi ||_{TV} \le || f_{y^{(0)}}^t - f ||_{TV}.
$$

$\Box$ 

Note that this setting contains as a particular case hidden Markov models, where $Y^{(t+1)} | x^{(t)}, y^{(t)} \sim f(y | y^{(t)})$ is not observed. These models will be detailed in Section 14.3.2.

\end{aligned}
$$

$\Box$

Unfortunately, this result has rather limited consequences for the study of the sequence $(X^{(t)})$ since (9.8) is an average property of $(X^{(t)})$, while MCMC algorithms such as [A.33] only produce a single realization from $(X^{(t)})$. We will see in Section 12.2.3 a more practical implication of the Duality Principle since Theorem 9.12 may allow a control of convergence by renewal.

```python
import random
import matplotlib.pyplot as plt

# 1. Define conditional distributions (replace with your actual distributions)
def pi_x_given_y(x, y):
    """Conditional distribution of X given Y (example)."""
    if y == 0:
        return [0.1 * (2.71828**(-((xi)**2) / 2)) for xi in x]  # Example: Normal-like distribution
    else:
        return [0.1 * (2.71828**(-((xi - 3)**2) / 2)) for xi in x] # Example: Shifted normal-like

def f_y_given_x_y(y_new, x, y_old, rho):  # Add rho parameter
    """Conditional distribution of Y, controlled by rho."""
    if y_old == 0:
        return rho if y_new == 0 else (1 - rho)
    else:
        return (1 - rho) if y_new == 0 else rho

# 2. Dual chain sampler with convergence tracking
def dual_chain_sampler(num_samples, y_initial, rho):
    x_samples = []
    y = y_initial
    x_range = [-5 + i * 0.1 for i in range(101)]
    x_means = []  # Track mean of x samples

    for t in range(num_samples):
        # Sample x given y
        probabilities = pi_x_given_y(x_range, y)
        total_prob = sum(probabilities)
        normalized_probs = [p / total_prob for p in probabilities]
        x = random.choices(x_range, weights=normalized_probs)[0]  # Sample using weights

        # Sample y given x and old y (using rho)
        y_new = random.choices([0, 1], weights=[f_y_given_x_y(0, x, y, rho), 1-f_y_given_x_y(0, x, y, rho)])[0]
        y = y_new
        x_samples.append(x)
        x_means.append(sum(x_samples) / len(x_samples)) # Calculate mean at each step
    return x_samples, x_means


# 3. Experiment with different rho values (convergence rates)
num_samples = 1000
y_initial = 0
rho_values = [0.1, 0.5, 0.9]  # Different convergence rates for Y chain

plt.figure(figsize=(10, 6))
for rho in rho_values:
    samples, x_means = dual_chain_sampler(num_samples, y_initial, rho)
    plt.plot(x_means, label=f"rho = {rho}")

plt.xlabel("Iteration")
plt.ylabel("Mean of X Samples")
plt.title("Convergence of X for Different Convergence Rates of Y")
plt.legend()
plt.grid(True)
plt.show()
```
