```python
import random
import math
import matplotlib.pyplot as plt

# Simulate Data (Example: Missing Data Model)
random.seed(42)  # For reproducibility
n_samples = 500
# True parameters (replace with your actual priors/knowledge)
mu_true = 0
sigma_true = 1
p_true = 0.7  # Probability of observing the data

# Generate complete data
x_complete = [random.normalvariate(mu_true, sigma_true) for _ in range(n_samples)]

# Introduce missingness (e.g., with probability 1-p_true)
x_observed = [xi if random.random() < p_true else None for xi in x_complete]


def two_stage_gibbs(x_observed, n_iter=10000):
    n_samples = len(x_observed)
    # Initialize missing values (e.g., impute with mean or random values)
    x_missing_imputed = [xi if xi is not None else random.normalvariate(0, 1) for xi in x_observed] # Simple imputation

    mu_samples = []
    sigma_samples = []

    for _ in range(n_iter):
        # Stage 1: Sample missing values
        for i in range(n_samples):
            if x_observed[i] is None:
                mu_current = sum(x_missing_imputed) / n_samples  # Current estimate of mu
                sigma_current = math.sqrt(sum((xi - mu_current)**2 for xi in x_missing_imputed) / (n_samples - 1)) if n_samples > 1 else 1 # Current estimate of sigma
                x_missing_imputed[i] = random.normalvariate(mu_current, sigma_current)

        # Stage 2: Sample parameters (mu and sigma)
        mu_current = sum(x_missing_imputed) / n_samples
        # Example: Inverse-gamma prior for sigma^2 (adapt as needed)
        # Assuming parameters for the inverse gamma prior.
        alpha = 2 # shape parameter
        beta = 2 # rate parameter
        sigma_sq_inv = random.gammavariate(alpha, beta) # Sample from a gamma distribution
        sigma_current = 1/math.sqrt(sigma_sq_inv)
        # Update x_missing_imputed with the new mu and sigma.
        for i in range(n_samples):
            if x_observed[i] is None:
                x_missing_imputed[i] = random.normalvariate(mu_current, sigma_current)

        mu_samples.append(mu_current)
        sigma_samples.append(sigma_current)

    return mu_samples, sigma_samples, x_missing_imputed


mu_samples, sigma_samples, x_filled = two_stage_gibbs(x_observed, n_iter=15000)

# Analyze Results (Example)
mu_est = sum(mu_samples[5000:]) / len(mu_samples[5000:])  # Discard burn-in
sigma_est = sum(sigma_samples[5000:]) / len(sigma_samples[5000:])

print(f"Estimated mu: {mu_est}")
print(f"Estimated sigma: {sigma_est}")

# Plot the samples (similar to Figure 9.1 - adapt as needed)
plt.figure(figsize=(8, 6))
plt.plot(mu_samples[5000:], sigma_samples[5000:], alpha=0.1)
plt.xlabel("mu")
plt.ylabel("sigma")
plt.title("Two-Stage Gibbs Sampler Samples")
plt.grid(True)
plt.show()

# Example: plot the filled data
plt.figure(figsize=(8, 6))
plt.hist(x_filled, bins=20, alpha=0.5, label='Filled Data')
plt.hist(x_complete, bins=20, alpha=0.5, label='Complete Data') # If you have it
plt.xlabel("x")
plt.title("Imputed Data Distribution")
plt.legend()
plt.show()
```

## Example 9.7: Grouped Counting Data and Gibbs Sampling

This notebook explores the analysis of grouped counting data using a Poisson model and a Gibbs sampling approach, as illustrated in Example 9.7.

### Data Description

Consider recording the number of passages of individuals per unit time past a sensor for 360 consecutive time units. The data is grouped, with the number of passages categorized into bins (0, 1, 2, 3, and 4 or more). The observed frequencies for each category are shown in Table 9.1.

| Number of Passages | 0   | 1   | 2   | 3   | 4 or more |
|--------------------|-----|-----|-----|-----|-----------|
| Number of Observations | 139 | 128 | 55  | 25  | 13        |

### Model and Likelihood

Assuming each observation follows a Poisson distribution $\mathcal{P}(\lambda)$, the likelihood function for the observed data is:

$$
l(\lambda | x_1, ..., x_5) \propto e^{-347\lambda} \lambda^{128 + 55 \times 2 + 25 \times 3} \left(1 - e^{-\lambda} \sum_{i=0}^{3} \frac{\lambda^i}{i!} \right)^{13}
$$

where $x_1 = 139, ..., x_5 = 13$ represent the observed frequencies.

### Gibbs Sampling Approach

To estimate the posterior distribution $\pi(\lambda | x_1, ..., x_5)$, we introduce latent variables $y = (y_1, ..., y_{13})$ representing the actual number of passages for the 13 observations in the "4 or more" category. We then consider the joint posterior distribution $\pi(\lambda, y_1, ..., y_{13} | x_1, ..., x_5)$ and employ a Gibbs sampler with two components: $\lambda$ and $y$.

**Algorithm A.35 - Poisson-Gamma Gibbs Sampler**

Given $\lambda^{(t-1)}$,

1. **Simulate** $Y_j^{(t)} \sim \mathcal{P}(\lambda^{(t-1)}) I_{y \geq 4}$ for $i = 1, ..., 13$

2. **Simulate** $\lambda^{(k)} \sim \mathcal{G}a(313 + \sum_{i=1}^{13} y_i^{(t)}, 360)$

### Rao-Blackwellized Estimator

The convergence of the Rao-Blackwellized estimator is analyzed:

$$
\delta_{rb} = \frac{1}{T} \sum_{t=1}^{T} \mathbb{E}[\lambda | x_1, ..., x_5, y_1^{(t)}, ..., y_{13}^{(t)}] = \frac{1}{360T} \sum_{t=1}^{T} \left(313 + \sum_{i=1}^{13} y_i^{(t)} \right)
$$

This estimator leverages the conditional distribution of $\lambda$ given the latent variables $y$ to improve estimation efficiency.



### Fig.2

Fig.2 (not included here) visually demonstrates the convergence of this Rao-Blackwellized estimator.

This example showcases how Gibbs sampling can be used to analyze grouped count data by introducing latent variables for censored observations and iteratively sampling from conditional distributions. The Rao-Blackwellized estimator provides an efficient way to estimate the parameter of interest.

```python
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import poisson, gamma  # For Poisson and Gamma distributions

# Data (from Table 9.1)
observations = [139, 128, 55, 25, 13]

def poisson_gamma_gibbs(observations, n_iter=10000):
    """
    Implements the Poisson-Gamma Gibbs sampler for grouped count data.

    Args:
        observations: A list of observed frequencies for each category.
        n_iter: The number of Gibbs sampling iterations.

    Returns:
        A tuple containing lists of sampled lambda values and the Rao-Blackwellized estimates.
    """

    n_total = sum(observations)  # Total number of time units (360 in this case)
    n_4plus = observations[-1]  # Number of observations in the "4 or more" category
    lambda_current = 1.0  # Initial value for lambda
    lambda_samples = []
    y_samples = []  # Store samples of y for Rao-Blackwellization
    rao_blackwellized_estimates = []

    for _ in range(n_iter):
        # 1. Sample y (latent variables)
        y_current = []
        for _ in range(n_4plus):
            while True:
                y_potential = poisson.rvs(lambda_current)  # Sample from Poisson
                if y_potential >= 4:
                    y_current.append(y_potential)
                    break

        # 2. Sample lambda
        alpha = 313 + sum(y_current)
        beta = n_total
        lambda_current = gamma.rvs(alpha, scale=1/beta)  # Use scipy.stats.gamma
        
        lambda_samples.append(lambda_current)
        y_samples.append(y_current)

        # Calculate Rao-Blackwellized estimate
        rao_blackwellized_estimate = (313 + sum(y_current)) / (360)
        rao_blackwellized_estimates.append(rao_blackwellized_estimate)

    return lambda_samples, rao_blackwellized_estimates


lambda_samples, rb_estimates = poisson_gamma_gibbs(observations, n_iter=15000)

# Analyze Results (Example)
lambda_est = sum(lambda_samples[5000:]) / len(lambda_samples[5000:])  # Discard burn-in

print(f"Estimated lambda: {lambda_est}")

# Plotting
plt.figure(figsize=(10, 5))

# Plot lambda samples
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(lambda_samples[5000:], alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("Lambda")
plt.title("Lambda Samples")
plt.grid(True)

# Plot Rao-Blackwellized estimates
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(rb_estimates[5000:], alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("Rao-Blackwellized Estimate")
plt.title("Rao-Blackwellized Estimates")
plt.grid(True)

plt.tight_layout() # Adjust subplot params so that subplots fit in to the figure area.
plt.show()


# Example: Histogram of lambda samples (after burn-in)
plt.hist(lambda_samples[5000:], bins=30, alpha=0.7)
plt.xlabel("Lambda")
plt.title("Histogram of Lambda Samples (Burn-in Discarded)")
plt.grid(True)
plt.show()
```

## Example .8: Grouped Multinomial Data

This example (from Tanner and Wong, 1987) considers a grouped multinomial model:

$$
X \sim \mathcal{M}_5(n; a_1\mu + b_1, a_2\mu + b_2, a_3\eta + b_3, a_4\eta + b_4, c(1 - \mu - \eta))
$$

where

$$
0 \le a_1 + a_2 = a_3 + a_4 = 1 - \sum_{i=1}^4 b_i = c \le 1,
$$

and the $a_i$, $b_i \ge 0$ are known constants based on genetic considerations.

**Equivalent Formulation:**

This model is equivalent to sampling from

$$
Y \sim \mathcal{M}_9(n; a_1\mu, b_1, a_2\mu, b_2, a_3\eta, b_3, a_4\eta, b_4, c(1 - \mu - \eta))
$$

with some observations aggregated (and thus missing):

$$
X_1 = Y_1 + Y_2, \quad X_2 = Y_3 + Y_4, \quad X_3 = Y_5 + Y_6, \quad X_4 = Y_7 + Y_8, \quad X_5 = Y_9.
$$

**Prior Distribution:**

A natural prior distribution for $(\mu, \eta)$ is the Dirichlet prior:

$$
\pi(\mu, \eta) \propto \mu^{\alpha_1 - 1} \eta^{\alpha_2 - 1} (1 - \mu - \eta)^{\alpha_3 - 1},
$$

where $\alpha_1, \alpha_2, \alpha_3$ are hyperparameters.

**Gibbs Sampling:**

To perform inference on $(\mu, \eta)$, a Gibbs sampler can be constructed.  This would involve iteratively sampling from the conditional distributions:

1. **Sample Y:** Given $(\mu, \eta)$, sample the unobserved counts Y from the multinomial distribution.  This step involves imputing the "missing" data due to the grouping.

2. **Sample Î¼:** Given Y and Î·, sample Î¼ from its conditional distribution. This will likely involve updating the parameters of the Dirichlet distribution based on the observed and imputed counts.

3. **Sample Î·:** Given Y and Î¼, sample Î· from its conditional distribution.  Similar to the previous step, this will involve updating the Dirichlet distribution parameters.

This example illustrates how Gibbs sampling can be applied to a more complex grouped data scenario with multinomial observations and two parameters of interest. The key idea is again to introduce latent variables (the unobserved counts Y) to facilitate sampling.

```python
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import dirichlet, multinomial  # For Dirichlet and Multinomial distributions

# Example Data and Constants (Replace with your actual data and values)
n = 100  # Total count
a1 = 0.5; a2 = 0.5; a3 = 0.5; a4 = 0.5  # Example values
b1 = 0.1; b2 = 0.1; b3 = 0.1; b4 = 0.1  # Example values
c = 1 - (b1 + b2 + b3 + b4)
alpha1 = 2; alpha2 = 2; alpha3 = 2  # Dirichlet prior hyperparameters
observed_counts = [20, 15, 18, 12, 35]  # Example observed counts (X1, X2, X3, X4, X5)

def grouped_multinomial_gibbs(observed_counts, n, a1, a2, a3, a4, b1, b2, b3, b4, c, alpha1, alpha2, alpha3, n_iter=10000):
    """
    Implements the Gibbs sampler for the grouped multinomial model.
    """

    mu_current = 0.3  # Initial values for mu and eta
    eta_current = 0.3
    mu_samples = []
    eta_samples = []

    for _ in range(n_iter):
        # 1. Sample Y (unobserved counts)
        # Calculate probabilities for the 9 categories based on current mu and eta
        probs = [a1 * mu_current, b1, a2 * mu_current, b2, a3 * eta_current, b3, a4 * eta_current, b4, c * (1 - mu_current - eta_current)]
        # Ensure probs sum to 1 (important due to floating point limitations)
        probs = [p / sum(probs) for p in probs]
        y_current = list(multinomial.rvs(n, probs)) # Use scipy.stats.multinomial

        # 2. Sample mu
        # Calculate the 'success' counts for mu (Y1 + Y3)
        success_mu = y_current[0] + y_current[2]
        # Calculate the 'failure' counts for mu (the remaining counts related to mu)
        failure_mu = n - success_mu

        alpha_mu = alpha1 + success_mu
        beta_mu = alpha2 + alpha3 + failure_mu # Combine the remaining alphas and counts as beta

        mu_current = dirichlet.rvs([alpha_mu, beta_mu])[0][0] # Use scipy.stats.dirichlet and sample only one value

        # 3. Sample eta (similar to mu)
        success_eta = y_current[4] + y_current[6]
        failure_eta = n - success_eta

        alpha_eta = alpha2 + success_eta
        beta_eta = alpha1 + alpha3 + failure_eta

        eta_current = dirichlet.rvs([alpha_eta, beta_eta])[0][0]

        mu_samples.append(mu_current)
        eta_samples.append(eta_current)

    return mu_samples, eta_samples


mu_samples, eta_samples = grouped_multinomial_gibbs(observed_counts, n, a1, a2, a3, a4, b1, b2, b3, b4, c, alpha1, alpha2, alpha3, n_iter=15000)

# Analyze Results (Example)
mu_est = sum(mu_samples[5000:]) / len(mu_samples[5000:])  # Discard burn-in
eta_est = sum(eta_samples[5000:]) / len(eta_samples[5000:])

print(f"Estimated mu: {mu_est}")
print(f"Estimated eta: {eta_est}")

# Plotting
plt.figure(figsize=(10, 5))

# Plot mu and eta samples
plt.subplot(1, 2, 1)
plt.plot(mu_samples[5000:], alpha=0.5, label = 'mu samples')
plt.xlabel("Iteration")
plt.ylabel("mu")
plt.title("Mu Samples")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(eta_samples[5000:], alpha=0.5, label = 'eta samples')
plt.xlabel("Iteration")
plt.ylabel("eta")
plt.title("Eta Samples")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

#Joint distribution of mu and eta
plt.figure(figsize = (8,6))
plt.scatter(mu_samples[5000:], eta_samples[5000:], alpha = 0.3, s = 10)
plt.xlabel("mu")
plt.ylabel("eta")
plt.title("Joint Distribution of mu and eta")
plt.grid(True)
plt.show()
```

## Example .9: Capture-Recapture Uniform Model

This example revisits the capture-recapture model from Examples 2.25 and 5.22.  The simplest case assumes the population size \(N\) is unknown, and each individual has a probability \(p\) of being captured in each experiment, regardless of past captures.

**Sufficient Statistic and Likelihood:**

For two successive captures, a sufficient statistic is the triplet $(n_{11}, n_{10}, n_{01})$, where $n_{11}$ is the number captured twice, $n_{10}$ is the number captured only in the first experiment, and $n_{01}$ is the number captured only in the second experiment.  With $n = n_{10} + n_{01} + 2n_{11}$ (the total number of captures), the likelihood is:

$$
l(N, p | n_{11}, n_{10}, n_{01}) \propto \binom{N}{n_{11}, n_{10}, n_{01}} p^n (1 - p)^{N - n}
$$

Castledine (1981) calls this a uniform likelihood.

**Factorization and Prior:**

The likelihood factors through $n$ and $n' = n_{10} + n_{11} + n_{01}$ (the total number of distinct individuals captured).  With a Poisson prior $\mathcal{P}(\lambda)$ on $N$ and a uniform prior $U_{[0,1]}$ on $p$, the posterior is:

$$
\pi(p, N | n, n') \propto \frac{N!}{(N - n')!} p^n (1 - p)^{N - n} \frac{e^{-\lambda} \lambda^N}{N!}
$$

This setup allows for Bayesian inference on $N$ and $p$ using the observed capture data.  Further analysis would typically involve deriving the full conditional distributions for $N$ and $p$ given the data and implementing a Gibbs sampler or another MCMC method.

##  Reversible and Interleaving Chains

This section introduces a key property of two-stage Gibbs samplers: **interleaving**.

**Definition .10: Interleaving Property**

Two Markov chains $(X^{(t)})$ and $(Y^{(t)})$ are said to be conjugate to each other with the interleaving property if:

(i) $X^{(t)}$ and $X^{(t+1)}$ are independent conditionally on $Y^{(t)}$;
(ii) $Y^{(t-1)}$ and $Y^{(t)}$ are independent conditionally on $X^{(t)}$; and
(iii) $(X^{(t)}, Y^{(t-1)})$ and $(X^{(t+1)}, Y^{(t)})$ are identically distributed under stationarity.

In simpler terms, interleaving means that each chain updates independently given the current state of the other chain, and the joint distribution of the chains has a specific symmetry across time steps.

**Lemma 9.11: Reversibility and Interleaving in Two-Stage Gibbs Samplers**

Each of the chains $(X^{(t)})$ and $(Y^{(t)})$ generated by a two-stage Gibbs sampling algorithm is reversible, and the chain $(X^{(t)}, Y^{(t)})$ satisfies the interleaving property.

**Proof (Sketch):**

The reversibility of each individual chain follows from the detailed balance condition (Theorem 6.46).  The interleaving property is a direct consequence of the construction of the two-stage Gibbs sampler, as the updates of each variable (or block of variables) are conditionally independent given the current values of the other variables.

The text notes that while the individual chains are reversible, the joint chain $(X^{(t)}, Y^{(t)})$ is not necessarily time-reversible.  This is an important distinction.

The interleaving property is a strong structural characteristic of two-stage Gibbs samplers, which has implications for their theoretical analysis and practical behavior.  It highlights the specific way in which the two "sub-chains" interact and update.

```python
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import beta, poisson  # For Beta and Poisson distributions

# Example Data and Constants (Replace with your actual data and values)
n11 = 25  # Number captured twice
n10 = 50  # Number captured only in the first experiment
n01 = 30  # Number captured only in the second experiment

n = n11 + n10 + n01  # Total captures
n_unique = n10 + n01 + n11 # Total unique individuals captured
lambda_poisson = 100  # Poisson prior parameter for N

def capture_recapture_gibbs(n11, n10, n01, lambda_poisson, n_iter=10000):
    """
    Implements the Gibbs sampler for the capture-recapture model.
    """

    N_current = n + 50  # Initial value for N (slightly larger than n)
    p_current = 0.5  # Initial value for p
    N_samples = []
    p_samples = []

    for _ in range(n_iter):
        # 1. Sample p
        alpha_p = n + 1
        beta_p = N_current - n + 1
        p_current = beta.rvs(alpha_p, beta_p) # Use scipy.stats.beta

        # 2. Sample N (Metropolis-Hastings step)
        while True:  # Sample N until a valid value is found
            N_proposal = poisson.rvs(lambda_poisson)  # Sample from Poisson prior
            if N_proposal >= n_unique: # Condition for N
                # Calculate the acceptance probability
                prior_ratio = (math.exp(-lambda_poisson) * lambda_poisson**N_proposal) / (math.exp(-lambda_poisson) * lambda_poisson**N_current)
                likelihood_ratio = (math.factorial(N_proposal) / math.factorial(N_proposal - n_unique)) / (math.factorial(N_current) / math.factorial(N_current - n_unique)) if N_proposal>=n_unique else 0 # Corrected likelihood ratio
                acceptance_prob = min(1, prior_ratio * likelihood_ratio)

                if random.random() < acceptance_prob:
                    N_current = N_proposal
                    break
        N_samples.append(N_current)
        p_samples.append(p_current)

    return N_samples, p_samples


N_samples, p_samples = capture_recapture_gibbs(n11, n10, n01, lambda_poisson, n_iter=15000)

# Analyze Results (Example)
N_est = sum(N_samples[5000:]) / len(N_samples[5000:])  # Discard burn-in
p_est = sum(p_samples[5000:]) / len(p_samples[5000:])

print(f"Estimated N: {N_est}")
print(f"Estimated p: {p_est}")

# Plotting
plt.figure(figsize=(10, 5))

# Plot N and p samples
plt.subplot(1, 2, 1)
plt.plot(N_samples[5000:], alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("N")
plt.title("N Samples")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(p_samples[5000:], alpha=0.5)
plt.xlabel("Iteration")
plt.ylabel("p")
plt.title("p Samples")
plt.grid(True)

plt.tight_layout()
plt.show()

#Joint distribution of N and p
plt.figure(figsize = (8,6))
plt.scatter(N_samples[5000:], p_samples[5000:], alpha = 0.3, s = 10)
plt.xlabel("N")
plt.ylabel("p")
plt.title("Joint Distribution of N and p")
plt.grid(True)
plt.show()
```
