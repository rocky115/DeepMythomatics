$$
\textbf{1.14 Credibility Estimates}

In actuarial studies, a credibility estimate is one which can be expressed as a weighted average of the form 

\[
C = (1 - k) A + kB,
\]

where:
\begin{itemize}
    \item \( A \) is the subjective estimate (or the collateral data estimate),
    \item \( B \) is the objective estimate (or the direct data estimate),
    \item \( k \) is the credibility factor, a number that is between 0 and 1 (inclusive) and represents the weight assigned to the objective estimate.
\end{itemize}

A high value of \( k \) implies \( C \approx B \), representing a situation where the objective estimate is assigned ‘high credibility’. A primary aim of credibility theory is to determine an appropriate value or formula for \( k \), as is done, for example, in the theory of the Bühlmann model (Bühlmann, 1967). Many Bayesian models lead to a point estimate which can be expressed as an intuitively appealing credibility estimate.

 # Exercise 1.16: Credibility Estimation in the Binomial-Beta Model

Consider the binomial-beta model given by:

$$
(y | \theta) \sim \text{Binomial}(n, \theta)
$$

$$
\theta \sim \text{Beta}(\alpha, \beta)
$$

## Posterior Distribution

We previously established that the posterior distribution of \(\theta\) given \(y\) is:

$$
(\theta | y) \sim \text{Beta}(\alpha + y, \beta + n - y)
$$

Thus, the posterior mean of \(\theta\) is:

$$
\hat{\theta} = E(\theta | y) = \frac{\alpha + y}{(\alpha + y) + (\beta + n - y)} = \frac{\alpha + y}{\alpha + \beta + n}
$$

## Prior Mean and Maximum Likelihood Estimate (MLE)

The prior mean of \(\theta\) is:

$$
E[\theta] = \frac{\alpha}{\alpha + \beta}
$$

The maximum likelihood estimate (MLE) of \(\theta\) is:

$$
\hat{\theta}_{MLE} = \frac{y}{n}
$$

## Expressing the Credibility Estimate

We can rewrite the posterior mean as:

$$
\hat{\theta} = \frac{\alpha}{\alpha + \beta + n} + \frac{n}{\alpha + \beta + n} \cdot \frac{y}{n}
$$

This leads us to the expression:

$$
\hat{\theta} = (1 - k) A + k B
$$

where:

- $A = E[\theta] = \frac{\alpha}{\alpha + \beta}$ (prior mean)
- $B = \frac{y}{n}$ (MLE)
- $k = \frac{n}{n + \alpha + \beta}$ (credibility factor)

## Discussion

The posterior mean $\hat{\theta}$ serves as a credibility estimate, expressed as a weighted average of the prior mean $A$ and the MLE $B$. The weight assigned to the MLE is determined by the credibility factor $k$, which reflects the amount of data available.

As \(n\) increases, the credibility factor $k$ approaches 1, indicating that the influence of the prior diminishes with increasing data. This aligns with our intuition: with sufficient data, the estimation should rely more on the observed outcomes rather than the prior beliefs.

### Exercise 6: Credibility Estimation in the Binomial-Beta Model

Consider the binomial-beta model given by:

$$
(y | \theta) \sim \text{Binomial}(n, \theta)
$$

$$
\theta \sim \text{Beta}(\alpha, \beta)
$$

## Posterior Distribution

We previously established that the posterior distribution of $\theta$ given $y$ is:

$$
(\theta | y) \sim \text{Beta}(\alpha + y, \beta + n - y)
$$

Thus, the posterior mean of $\theta$ is:

$$
\hat{\theta} = E(\theta | y) = \frac{\alpha + y}{(\alpha + y) + (\beta + n - y)} = \frac{\alpha + y}{\alpha + \beta + n}
$$

## Prior Mean and Maximum Likelihood Estimate (MLE)

The prior mean of $\theta$ is:

$$
E[\theta] = \frac{\alpha}{\alpha + \beta}
$$

The maximum likelihood estimate (MLE) of $\theta$ is:

$$
\hat{\theta}_{MLE} = \frac{y}{n}
$$

## Case (a): $n = 5$, $y = 4$, $\alpha = 2$, $\beta = 6$

In this case, we compute the following:

- Prior Mean $A$:

$$
A = \frac{\alpha}{\alpha + \beta} = \frac{2}{2 + 6} = 0.25
$$

- MLE \(B\):

$$
B = \frac{y}{n} = \frac{4}{5} = 0.8
$$

- Posterior Mean:

The posterior parameters are:

$$
\hat{\theta} = \frac{\alpha + y}{\alpha + \beta + n} = \frac{2 + 4}{2 + 6 + 5} = \frac{6}{13} \approx 0.462
$$

- Credibility Factor \(k\):

$$
k = \frac{n}{n + \alpha + \beta} = \frac{5}{5 + 2 + 6} = \frac{5}{13} \approx 0.385
$$

## Case (b): $n = 20$, $y = 16$, $\alpha = 2$, $\beta = 6$

For this case, we compute:

- Prior Mean \(A\):

$$
A = \frac{2}{2 + 6} = 0.25
$$

- MLE \(B\):

$$
B = \frac{y}{n} = \frac{16}{20} = 0.8
$$

- Posterior Mean:

The posterior parameters are:

$$
\hat{\theta} = \frac{\alpha + y}{\alpha + \beta + n} = \frac{2 + 16}{2 + 6 + 20} = \frac{18}{28} \approx 0.643
$$

- Credibility Factor $k$:

$$
k = \frac{n}{n + \alpha + \beta} = \frac{20}{20 + 2 + 6} = \frac{20}{28} \approx 0.714
$$

## Comparison of Cases

Both cases have the same prior mean and MLE:

- **Prior Mean:** $A = 0.25$
- **MLE:** $B = 0.8$

However, due to the larger sample size $n$ in case (b), the credibility factor is larger:

- **Credibility Factor:** $k \approx 0.714$ (case b) vs $k \approx 0.385$ (case a)

This leads to a posterior mean closer to the MLE in case (b):

- **Posterior Mean:** $\hat{\theta} \approx 0.643$ (case b) vs $\hat{\theta} \approx 0.462$ (case a)

## Note on Likelihood Functions

Each likelihood function in Figure 1.9 has been normalized so that the area underneath it is exactly 1. This means that in each case (a) and (b), the likelihood function \(L(\theta)\) as shown is identical to the posterior density implied by the standard uniform prior, i.e. under:

$$
f_U(0, 1)(\theta) = f_{\text{Beta}}(1, 1)(\theta)
$$

Thus, we have:

$$
L(\theta) = f_{\text{Beta}}(1 + y, 1 + n - y)(\theta)
$$

Figure 9 illustrates the concept of credibility estimation by showing the relevant densities, likelihoods, and estimates for varying data cases.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Function to compute posterior mean and credibility factor
def credibility_estimate(alpha, beta_param, n, y):
    # Prior mean
    A = alpha / (alpha + beta_param)
    
    # MLE
    B = y / n
    
    # Posterior parameters
    alpha_post = alpha + y
    beta_post = beta_param + (n - y)
    
    # Posterior mean
    theta_hat = alpha_post / (alpha_post + beta_post)
    
    # Credibility factor
    k = n / (n + alpha + beta_param)
    
    return A, B, theta_hat, k, alpha_post, beta_post

# Parameters for Case (a)
n_a = 5
y_a = 4
alpha_a = 2
beta_a = 6

# Parameters for Case (b)
n_b = 20
y_b = 16
alpha_b = 2
beta_b = 6

# Calculate for both cases
results_a = credibility_estimate(alpha_a, beta_a, n_a, y_a)
results_b = credibility_estimate(alpha_b, beta_b, n_b, y_b)

# Display results
print("Case (a):")
print(f"Prior Mean (A): {results_a[0]:.4f}, MLE (B): {results_a[1]:.4f}, Posterior Mean: {results_a[2]:.4f}, Credibility Factor (k): {results_a[3]:.4f}")
print("\nCase (b):")
print(f"Prior Mean (A): {results_b[0]:.4f}, MLE (B): {results_b[1]:.4f}, Posterior Mean: {results_b[2]:.4f}, Credibility Factor (k): {results_b[3]:.4f}")

# Plotting the likelihood and posterior distributions for both cases
theta = np.linspace(0, 1, 100)

# Case (a) likelihood and posterior
likelihood_a = (theta ** y_a) * ((1 - theta) ** (n_a - y_a))
posterior_a = beta.pdf(theta, results_a[4], results_a[5])

# Case (b) likelihood and posterior
likelihood_b = (theta ** y_b) * ((1 - theta) ** (n_b - y_b))
posterior_b = beta.pdf(theta, results_b[4], results_b[5])

# Normalize likelihoods
likelihood_a /= np.trapz(likelihood_a, theta)
likelihood_b /= np.trapz(likelihood_b, theta)

# Plotting
plt.figure(figsize=(14, 6))

# Case (a)
plt.subplot(1, 2, 1)
plt.plot(theta, likelihood_a, label='Likelihood (Case a)', color='blue')
plt.plot(theta, posterior_a, label='Posterior (Case a)', color='orange')
plt.axvline(results_a[2], color='red', linestyle='--', label='Posterior Mean (Case a)')
plt.title('Case (a): n=5, y=4, α=2, β=6')
plt.xlabel('θ')
plt.ylabel('Density')
plt.legend()
plt.grid()

# Case (b)
plt.subplot(1, 2, 2)
plt.plot(theta, likelihood_b, label='Likelihood (Case b)', color='blue')
plt.plot(theta, posterior_b, label='Posterior (Case b)', color='orange')
plt.axvline(results_b[2], color='red', linestyle='--', label='Posterior Mean (Case b)')
plt.title('Case (b): n=20, y=16, α=2, β=6')
plt.xlabel('θ')
plt.ylabel('Density')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

## Exercise 7: Further Credibility Estimation in the Binomial-Beta Model

Consider the binomial-beta model given by:

$$
(Y | \theta) \sim \text{Binomial}(n, \theta)
$$

$$
\theta \sim \text{Beta}(\alpha, \beta)
$$

## Posterior Distribution

From previous analysis, we know that the posterior distribution of $\theta$ given $y$ is:

$$
(\theta | y) \sim \text{Beta}(\alpha + y, \beta + n - y)
$$

## Posterior Mode

The posterior mode of \(\theta\) can be expressed as:

$$
\text{Mode}(\theta | y) = \frac{\alpha + y - 1}{(\alpha + y - 1) + (\beta + n - y - 1)} = \frac{\alpha + y - 1}{\alpha + \beta + n - 2}
$$

## Prior Mode

The prior mode of $\theta$ is given by:

$$
\text{Mode}(\theta) = \frac{\alpha - 1}{(\alpha - 1) + (\beta - 1)} = \frac{\alpha - 1}{\alpha + \beta - 2}
$$

## Expressing the Posterior Mode

We can rewrite the posterior mode as:

$$
\text{Mode}(\theta | y) = \frac{\alpha - 1}{\alpha + \beta - 2} + \frac{n}{\alpha + \beta + n - 2} \cdot \frac{y}{n}
$$

This leads us to the form:

$$
\text{Mode}(\theta | y) = (1 - c) \cdot \text{Mode}(\theta) + c \cdot \hat{\theta}
$$

where:

- \$\text{Mode}(\theta) = \frac{\alpha - 1}{\alpha + \beta - 2} \quad \text{(prior mode)}$
- $\hat{\theta} = \frac{y}{n} \quad \text{(maximum likelihood estimate)}$
- $c = \frac{n}{n + \alpha + \beta - 2} \quad \text{(credibility factor)}$

## Discussion

The posterior mode $\text{Mode}(\theta | y)$ serves as a credibility estimate expressed as a weighted average of the prior mode and the MLE, where the weight assigned to the MLE is determined by the credibility factor $c$. This reflects how much influence the observed data has on the estimation compared to the prior belief. As $n$ increases, the credibility factor $c$ approaches 1, meaning the estimate becomes more reliant on the observed data.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Function to compute posterior mode and credibility factor
def credibility_mode(alpha, beta_param, n, y):
    # Prior mode
    if alpha > 1:
        prior_mode = (alpha - 1) / (alpha + beta_param - 2)
    else:
        prior_mode = 0  # Undefined for beta < 1
    
    # MLE
    mle = y / n
    
    # Posterior parameters
    alpha_post = alpha + y
    beta_post = beta_param + (n - y)
    
    # Posterior mode
    if alpha_post > 1:
        posterior_mode = (alpha_post - 1) / (alpha_post + beta_post - 2)
    else:
        posterior_mode = 0  # Undefined for beta < 1
    
    # Credibility factor
    credibility_factor = n / (n + alpha + beta_param - 2)
    
    return prior_mode, mle, posterior_mode, credibility_factor

# Parameters for Case (a)
n_a = 5
y_a = 4
alpha_a = 2
beta_a = 6

# Parameters for Case (b)
n_b = 20
y_b = 16
alpha_b = 2
beta_b = 6

# Calculate for both cases
results_a = credibility_mode(alpha_a, beta_a, n_a, y_a)
results_b = credibility_mode(alpha_b, beta_b, n_b, y_b)

# Display results
print("Case (a):")
print(f"Prior Mode: {results_a[0]:.4f}, MLE: {results_a[1]:.4f}, Posterior Mode: {results_a[2]:.4f}, Credibility Factor: {results_a[3]:.4f}")
print("\nCase (b):")
print(f"Prior Mode: {results_b[0]:.4f}, MLE: {results_b[1]:.4f}, Posterior Mode: {results_b[2]:.4f}, Credibility Factor: {results_b[3]:.4f}")

# Plotting the likelihood and posterior distributions for both cases
theta = np.linspace(0, 1, 100)

# Case (a) likelihood and posterior
likelihood_a = (theta ** y_a) * ((1 - theta) ** (n_a - y_a))
posterior_a = beta.pdf(theta, results_a[0] * (alpha_a + beta_a - 2), (1 - results_a[0]) * (alpha_a + beta_a - 2))

# Case (b) likelihood and posterior
likelihood_b = (theta ** y_b) * ((1 - theta) ** (n_b - y_b))
posterior_b = beta.pdf(theta, results_b[0] * (alpha_b + beta_b - 2), (1 - results_b[0]) * (alpha_b + beta_b - 2))

# Normalize likelihoods
likelihood_a /= np.trapz(likelihood_a, theta)
likelihood_b /= np.trapz(likelihood_b, theta)

# Plotting
plt.figure(figsize=(14, 6))

# Case (a)
plt.subplot(1, 2, 1)
plt.plot(theta, likelihood_a, label='Likelihood (Case a)', color='blue')
plt.plot(theta, posterior_a, label='Posterior (Case a)', color='orange')
plt.axvline(results_a[2], color='red', linestyle='--', label='Posterior Mode (Case a)')
plt.title('Case (a): n=5, y=4, α=2, β=6')
plt.xlabel('θ')
plt.ylabel('Density')
plt.legend()
plt.grid()

# Case (b)
plt.subplot(1, 2, 2)
plt.plot(theta, likelihood_b, label='Likelihood (Case b)', color='blue')
plt.plot(theta, posterior_b, label='Posterior (Case b)', color='orange')
plt.axvline(results_b[2], color='red', linestyle='--', label='Posterior Mode (Case b)')
plt.title('Case (b): n=20, y=16, α=2, β=6')
plt.xlabel('θ')
plt.ylabel('Density')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

## Exercise 8: The Normal-Normal Model

Consider the following Bayesian model:

$$(y_1, \ldots, y_n | \mu) \sim \text{iid } N(\mu, \sigma^2)$$
$$\mu \sim N(\mu_0, \sigma_0^2$$

where $\sigma^2$, $\mu_0$, and $\sigma_0^2$ are known or specified constants.

### Posterior Distribution

The posterior density of $\mu$ is given by:

$$ f(\mu | y) \propto f(\mu) f(y | \mu) $$

The likelihood can be expressed as:

$$ f(y | \mu) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y_i - \mu)^2}{2\sigma^2}\right) $$

The prior density function is:

$$ f(\mu) = \frac{1}{\sqrt{2\pi\sigma_0^2}} \exp\left(-\frac{(\mu - \mu_0)^2}{2\sigma_0^2}\right) $$

Combining these, we have:

$$
f(\mu | y) \propto \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (y_i - \mu)^2\right) \cdot \exp\left(-\frac{(\mu - \mu_0)^2}{2\sigma_0^2}\right)
$$

This simplifies to:

$$
f(\mu | y) \propto \exp\left(-\frac{1}{2}\left[\frac{1}{\sigma^2} \sum_{i=1}^{n} (y_i - \mu)^2 + \frac{(\mu - \mu_0)^2}{\sigma_0^2}\right]\right)
$$

## Completing the Square

The posterior density of $\mu$ can be expressed in a quadratic form. We denote:

$$
q = \frac{1}{2}\left( \left(\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}\right)\mu^2 - 2\left(\frac{ny}{\sigma^2} + \frac{\mu_0}{\sigma_0^2}\right)\mu + c\right)
$$

where $c$ is a constant that does not depend on $\mu$.

To complete the square, we can rewrite this as:

$$
q = a\mu^2 - 2b\mu + c
$$

where

$$
a = \frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}, \quad b = \frac{ny}{\sigma^2} + \frac{\mu_0}{\sigma_0^2}
$$

This gives us:

$$
q = a\left(\mu - \frac{b}{a}\right)^2 + d
$$

where $d$ is another constant term.

## Posterior Distribution

Thus, the posterior distribution is:

$$
f(\mu | y) \sim N(\mu^*, \sigma^{*2})
$$

Where:

- The posterior mean:

$$
\mu^* = \frac{\frac{ny}{\sigma^2} + \frac{\mu_0}{\sigma_0^2}}{\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}}
$$

- The posterior variance:

$$
\sigma^{*2} = \frac{1}{\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}}
$$


## Additional Notes on the Normal-Normal Model

**Note 1:** A little algebra (left as an additional exercise) shows that the posterior mean can also be written as:

$$
\mu^* = (1 - k)\mu_0 + ky,
$$

and the posterior variance can be written as:

$$
\sigma^{*2} = \frac{\sigma^2}{k},
$$

where 

$$
k = \frac{n}{n + \frac{2\sigma_0^2}{\sigma^2}}.
$$

We see that $\mu$’s posterior mean is a credibility estimate in the form of a weighted average of the prior mean $\mu_0$ and the sample mean $y$ (which is also the maximum likelihood estimate), with the weight assigned to $y$ being the credibility factor $k$.

---

**Note 2:** Another way to derive $\mu^*$ and $\sigma^{*2}$ is to write (1.2) as:

$$
f(\mu | y) \propto \exp\left(-\frac{1}{2} \left(\mu^2 - 2\mu\mu^* + \mu^{*2}\right)\right)
$$

and then equate coefficients of powers of $\mu$ in (1.1) and (1.5). This logic leads to:

$$
\frac{1}{\sigma^2} + \frac{1}{\sigma_0^2} = \frac{1}{\sigma^{*2}}, \quad \text{and} \quad \mu^* = \frac{\frac{ny}{\sigma^2} + \frac{\mu_0}{\sigma_0^2}}{\frac{n}{\sigma^2} + \frac{1}{\sigma_0^2}}.
$$

This results in the same formulae for $\mu^*$ and $\sigma^{*2}$ as given by (1.4).

---

**Note 3:** Since both prior and posterior are normal, the prior is conjugate.

---

**Note 4:** The posterior mean, mode, and median of $\mu$ are the same and equal to $\mu^*$. The $1 - \alpha$ Credible Posterior Distribution Region (CPDR) and $1 - \alpha$ Highest Posterior Density Region (HPDR) for $\mu$ are the same and equal to:

$$
\left(\mu^* \pm z_{\frac{\alpha}{2}} \sigma^{*}\right).
$$

---

**Note 5:** The posterior distribution of $\mu$ depends on the data vector $y = (y_1, \ldots, y_n)$ only by way of the sample mean, i.e., 

$$
y = \frac{y_1 + \cdots + y_n}{n}.
$$

Therefore, the main result $(\mu | y) \sim N(\mu^*, \sigma^{*2})$ also implies that $(\mu | y) \sim N(\mu^*, \sigma^2)$. That is, if we know only the sample mean $y$, the posterior distribution of $\mu$ is the same as if we know $y$, i.e., all $n$ sample values. Knowing the individual $y_i$ values makes no difference to the inference.

---

**Note 6:** The formula for the credibility factor from Note 1, namely:

$$
k = \frac{n}{\frac{\sigma^2}{n} + 2\sigma_0^2}
$$

makes sense in the following ways:

1. If the prior standard deviation $\sigma_0$ is small, then $k \approx 0$, so that $\mu^* \approx \mu_0$ and $\sigma^{*} \approx \sigma_0$. Therefore, $(\mu | y) \sim N(\mu_0, \sigma_0^2)$. In this case, the posterior mean, mode, and median of $\mu$ are approximately equal to $\mu_0$. Also, the $1 - \alpha$ CPDR and $1 - \alpha$ HPDR for $\mu$ are approximately equal to:

$$
\left(\mu_0 \pm z_{\frac{\alpha}{2}} \sigma_0\right).
$$

2. If $\sigma_0$ is large, then $k \approx 1$, so that $\mu^* \approx y$ and $\sigma^{*2} \approx \frac{\sigma^2}{n}$, and thus $(\mu | y) \sim N\left(y, \frac{\sigma^2}{n}\right)$.


## Additional Notes on the Normal-Normal Model

That is, a large $\sigma_0$ corresponds to a highly disperse prior, reflecting little prior information and so little influence of the prior distribution (as specified by $\mu_0$ and $\sigma_0$) on the inference. In this case, the posterior mean, mode, and median of $\mu$ are approximately equal to $y$. Also, the $1 - \alpha$ CPDR and $1 - \alpha$ HPDR for $\mu$ are approximately equal to:

$$
\left(y \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}\right).
$$

Thus, inference is almost the same as implied by the classical approach.

---

If the sample size $n$ is large, then $k \approx 1$, so that $\mu^* \approx y$ and $\sigma^{*2} \approx \frac{\sigma^2}{n}$. Therefore, 

$$
(\mu | y) \sim N\left(y, \frac{\sigma^2}{n}\right).
$$

In this case, just as when $\sigma_0$ is large, the prior distribution has very little influence on the posterior, and the ensuing inference is almost the same as that implied by the classical approach.

---

**Note 7:** In the case of a priori ignorance (meaning no prior information at all), it is customary to take $\sigma_0 = \infty$, which implies that 

$$
\mu \sim N(0, \infty).
$$

This prior on $\mu$ appears to be problematic because it is improper. However, it meaningfully leads to a proper posterior:

$$
(\mu | y) \sim N\left(y, \frac{\sigma^2}{n}\right),
$$

which then leads to the same point and interval estimates implied by the classical approach, namely the MLE $y$ and the $1 - \alpha$ confidence interval 

$$
\left(y \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}\right).
$$

The improper prior $\mu \sim N(0, \infty)$ may be described as ‘flat’ or ‘uniform over the whole real line’ and can also be written as $\mu \sim U(-\infty, \infty)$ or $f(\mu) \propto 1, \ \mu \in \mathbb{R}$. In some cases (more complicated models not considered here), using an improper prior may lead to an improper posterior, which then becomes problematic. For more information on this topic, see Hobert and Casella (1996).

**Summary:** For the normal-normal model, defined by:

$$
(y_1, \ldots, y_n | \mu) \sim \text{iid} \ N(\mu, \sigma^2), \quad \mu \sim N(\mu_0, \sigma_0^2),
$$

the posterior distribution of the normal mean $\mu$ is given by:

$$
(\mu | y) \sim N(\mu^*, \sigma^{*2}),
$$

where:

$$
\mu^* = (1 - k)\mu_0 + ky,
$$
$$
\sigma^{*2} = \frac{\sigma^2}{k},
$$
$$
k = \frac{n}{n + \frac{\sigma^2}{\sigma_0^2}} \quad \text{(the normal-normal model credibility factor)}.
$$

The posterior mean, mode, and median of $\mu$ are all equal to $\mu^*$, and the $1 - \alpha$ CPDR and HPDR for $\mu$ are both:

$$
\left(\mu^* \pm z_{\frac{\alpha}{2}} \sigma^*\right).
$$

In the case of a priori ignorance, it is appropriate to set $\sigma_0 = \infty$. This defines an improper prior $f(\mu) \propto 1, \ \mu \in \mathbb{R}$ and the proper posterior:

$$
(\mu | y) \sim N\left(y, \frac{\sigma^2}{n}\right).
$$


## Summary of the Normal-Normal Model

For the normal-normal model, defined by:

$$
(y_1, \ldots, y_n | \mu) \sim \text{iid} \ N(\mu, \sigma^2), \quad \mu \sim N(\mu_0, \sigma_0^2),
$$

$$

k = \frac{n}{n + \frac{\sigma^2}{\sigma_0^2}} \quad \text{(the normal-normal model credibility factor)}.
$$

The posterior mean, mode, and median of $\mu$ are all equal to $\mu^*$, and the $1 - \alpha$ CPDR and HPDR for $\mu$ are both:

$$
\left(\mu^* \pm z_{\frac{\alpha}{2}} \sigma^*\right).
$$

In the case of a priori ignorance, it is appropriate to set $\sigma_0 = \infty$. This defines an improper prior:

$$
f(\mu) \propto 1, \quad \mu \in \mathbb{R},
$$

and the proper posterior is:

$$
(\mu | y) \sim N\left(y, \frac{\sigma^2}{n}\right).
$$


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the parameters
mu_0 = 0        # Prior mean
sigma_0_sq = 1  # Prior variance
sigma_sq = 1    # Likelihood variance
y = np.array([1.2, 1.5, 1.8, 2.0])  # Sample data
n = len(y)     # Sample size
y_bar = np.mean(y)  # Sample mean

# Calculate the posterior parameters
k = n / (n + sigma_sq / sigma_0_sq)  # Credibility factor
mu_star = (1 - k) * mu_0 + k * y_bar  # Posterior mean
sigma_star_sq = sigma_sq / k          # Posterior variance

# Posterior distribution
posterior_distribution = stats.norm(mu_star, np.sqrt(sigma_star_sq))

# Create a range of values for plotting
x = np.linspace(mu_star - 4 * np.sqrt(sigma_star_sq), mu_star + 4 * np.sqrt(sigma_star_sq), 100)
y_pdf = posterior_distribution.pdf(x)

# Plotting the posterior distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y_pdf, label='Posterior Distribution', color='blue')
plt.axvline(mu_star, color='red', linestyle='--', label='Posterior Mean')
plt.title('Posterior Distribution of μ')
plt.xlabel('μ')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()

# Print the results
print(f'Posterior Mean (μ*): {mu_star}')
print(f'Posterior Variance (σ*²): {sigma_star_sq}')
```

# Exercise 9: Practice with the Normal-Normal Model

In the context of the normal-normal model, defined by:

$$
(y_1, \ldots, y_n | \mu) \sim \text{iid } N(\mu, \sigma^2), \quad \mu \sim N(\mu_0, \sigma_0^2)
$$

we have the following values:

- Sample data: $y = (8.4, 10.1, 9.4)$
- Known variance: $\sigma = 1$
- Prior mean: $\mu_0 = 5$
- Prior variance: $\sigma_0 = \frac{1}{2}$

## Step 1: Calculate the Sample Mean

The sample mean $\bar{y}$ is calculated as follows:

$$
\bar{y} = \frac{8.4 + 10.1 + 9.4}{3} = 9.3
$$

## Step 2: Calculate Sample Size

The sample size \(n\) is given by:

$$
n = 3
$$

## Step 3: Calculate the Credibility Factor \(k\)

The credibility factor \(k\) can be calculated using the formula:

$$
k = \frac{n}{n + \frac{\sigma^2}{\sigma_0^2}} = \frac{3}{3 + \frac{1^2}{(1/2)^2}} = \frac{3}{7} \approx 0.4286
$$

## Step 4: Calculate the Posterior Mean $\mu^*$

The posterior mean \(\mu^*\) is given by:

$$
\mu^* = (1 - k) \mu_0 + k \bar{y} = \left(1 - \frac{3}{7}\right) \cdot 5 + \frac{3}{7} \cdot 9.3 \approx 6.84286
$$

## Step 5: Calculate the Posterior Variance \(\sigma^{*2}\)

The posterior variance \(\sigma^{*2}\) is calculated as:

$$
\sigma^{*2} = \frac{\sigma^2}{k} = \frac{1^2}{\frac{3}{7}} = \frac{7}{3} \approx 0.142857
$$

## Step 6: Calculate the 95% CPDR and HPDR

The 95% credible and highest posterior density regions (CPDR/HPDR) are:

$$
\text{95% CPDR/HPDR} = \left(\mu^* \pm z_{0.025} \sigma^*\right) = \left(6.84286 \pm 1.96 \sqrt{0.142857}\right)
$$

Calculating the bounds:

$$
= \left(6.84286 \pm 1.96 \cdot 0.377964\right) = (6.10256, 7.58316)
$$

## Results

- **Posterior Mean $(\mu^*$)**: $6.84286$
- **Posterior Variance $(\sigma^{*2}$)**: $0.14286$
- **95% CPDR**: $(6.10256, 7.58316)$
