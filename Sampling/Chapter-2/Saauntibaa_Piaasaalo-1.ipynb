## Bayesian Breeze

### Frequentist characteristics of Bayesian estimators

Consider a Bayesian model defined by a likelihood $ f(y|\theta) $ and a prior $ f(\theta) $, leading to the posterior

$$
f(\theta|y) = \frac{f(\theta) f(y|\theta)}{f(y)}
$$

Suppose that we choose to perform inference on $ \theta $ by constructing a point estimate $ \hat{\theta} $ (such as the posterior mean, mode, or median) and a $ (1 - \alpha) $-level interval estimate $ I = (L, U) $ (such as the CPDR or HPDR). Then $ \hat{\theta}, I, L, $ and $ U $ are functions of the data $ y $ and may be written as $ \hat{\theta}(y), I(y), L(y), $ and $ U(y) $.

Once these functions are defined, the estimates which they define stand on their own, so to speak, and may be studied from many different perspectives. Naturally, the characteristics of these estimates may be seen in the context of the Bayesian framework in which they were constructed. More will be said on this below when we come to discuss Bayesian decision theory.

However, another important use of Bayesian estimates is as a proxy for classical estimates. We have already mentioned this in relation to the normal-normal model:

$$
(y_1, \dots, y_n | \mu) \sim \text{iid } N(\mu, \sigma^2)
$$

$$
\mu \sim N(\mu_0, \sigma_0^2)
$$

where the use of a particular prior, namely the one specified by $ \sigma_0 = \infty $, led to the point estimate:

$$
\hat{\mu} = \hat{\mu}(y) = \bar{y}
$$

and the interval estimate

$$
I(y) = (L(y), U(y)) = \left( \bar{y} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \right)
$$

As we noted earlier, these estimates are exactly the same as the usual estimates used in the context of the corresponding classical model.
$y_1, \dots, y_n \sim \text{iid } N(\mu, \sigma^2)$, where $ \mu $ is an unknown constant and $ \sigma^2 $ is given. Therefore, the frequentist operating characteristics of the Bayesian estimates are immediately known.

In particular, we refer to the fact that the frequentist bias of $ \hat{\mu} $ is zero, and the frequentist coverage probability of $ I $ is exactly $ 1 - \alpha $. These statements mean that the expected value of $ y $ given $ \mu $ is $ \mu $ for all possible values of $ \mu $, and that the probability of $ \mu $ being inside $ I $ given $ \mu $ is $ 1 - \alpha $ for all possible values of $ \mu $.

More generally, in the context of a Bayesian model as above, we may define the frequentist bias of a Bayesian point estimate $ \hat{\theta} = \hat{\theta}(y) $ as:

$$
B_{\theta} = E\{\hat{\theta}(y) - \theta | \theta \}
$$

We may define the frequentist relative bias of $ \hat{\theta} $ as:

$$
R_{\theta} = E \left( \frac{\hat{\theta}(y) - \theta}{\theta} \right) = \frac{B_{\theta}}{\theta} \quad (\theta \neq 0)
$$

Furthermore, we define the frequentist coverage probability (FCP) of a Bayesian interval estimate $ I(y) = (L(y), U(y)) $ as:

$$
C_{\theta} = P\{\theta \in I(y) | \theta\}
$$

Thus, for the normal-normal model with \( \sigma_0 = \infty \), we may write:

$$
B_{\mu} = E\{\hat{\mu}(y) - \mu | \mu\} = E(y | \mu) - \mu = \mu - \mu = 0 \quad \forall \mu \in \mathbb{R}
$$

$$
R_{\mu} = \frac{B_{\mu}}{\mu} = 0 \quad (\mu \neq 0)
$$

The frequentist coverage probability is:

$$
C_{\mu} = P\{\mu \in I(y) | \mu\} = P\left( \frac{y - z_{\alpha/2}\frac{\sigma}{\sqrt{n}} < \mu < y + z_{\alpha/2}\frac{\sigma}{\sqrt{n}}}{\sigma/\sqrt{n}} \right) = 1 - \alpha \quad \forall \mu \in \mathbb{R}
$$

The above analysis is straightforward enough. However, in the case of an informative prior (one with $ \sigma_0 < \infty $), or in the context of other models...


### 1. Frequentist Bias of Bayesian Estimates

The **frequentist bias** of a Bayesian point estimate $ \hat{\theta}(y) $ is defined as:

$$
B_{\theta} = E\left[\hat{\theta}(y) - \theta \mid \theta \right]
$$

If $ B_{\theta} = 0 $, the estimator is unbiased in a frequentist sense. A non-zero bias indicates that the Bayesian estimate deviates from the true parameter $ \theta $.



### Bayesian Models and Frequentist Characteristics

In Bayesian models, the frequentist bias of a Bayesian point estimate $ B_{\theta} $ and the frequentist coverage probability $ C_{\theta} $ of a Bayesian interval estimate may not always be straightforward to compute or interpret. However, they are critical for understanding how well Bayesian estimates perform in a frequentist sense.

#### 1. Frequentist Bias of Bayesian Estimates

The **frequentist bias** of a Bayesian point estimate $ \hat{\theta}(y) $ measures the expected difference between the Bayesian estimate and the true parameter $ \theta $ across repeated samples of the data $ y $. It is defined as:

$$
B_{\theta} = E\left[\hat{\theta}(y) - \theta \mid \theta \right]
$$

If $ B_{\theta} = 0 $, then the estimate is unbiased from a frequentist perspective, meaning that the expected value of the Bayesian estimator equals the true parameter $ \theta $. If the bias is non-zero, the estimate is biased in a frequentist sense. Understanding this bias helps determine whether the Bayesian estimate can serve as a good proxy for a frequentist estimator.

#### 2. Frequentist Relative Bias

The **frequentist relative bias** $ R_{\theta} $ normalizes the absolute bias by the true parameter value, showing the relative size of the bias compared to the magnitude of $ \theta $. It is defined as:

$$
R_{\theta} = \frac{B_{\theta}}{\theta} = E\left[\frac{\hat{\theta}(y) - \theta}{\theta} \mid \theta \right] \quad (\theta \neq 0)
$$

This is particularly useful when comparing the bias of estimators across different scales or magnitudes of the parameter $ \theta $.

#### 3. Frequentist Coverage Probability of Bayesian Interval Estimates

The **frequentist coverage probability** (FCP) $ C_{\theta} $ measures how often a Bayesian interval estimate $ I(y) = (L(y), U(y)) $ contains the true parameter $ \theta $ in repeated samples of $ y $. It is defined as:

$$
C_{\theta} = P\left[\theta \in I(y) \mid \theta \right]
$$

If $ C_{\theta} $ is close to the nominal coverage (e.g., 95%), then the Bayesian interval performs well in a frequentist sense. If the coverage is too low, the interval may be too narrow, leading to undercoverage, while overly high coverage might indicate an unnecessarily wide interval.

#### 4. Applications in Decision-Making

Evaluating $ B_{\theta} $ and $ C_{\theta} $ helps us understand whether the Bayesian estimates can be used as frequentist proxies, especially when prior information is incorporated into the model. For example, if a Bayesian point estimate has a low frequentist bias and its interval estimates provide good frequentist coverage, we can use those Bayesian estimates as reliable frequentist estimates.

#### 5. Informative Priors

The situation becomes more nuanced when dealing with informative priors (e.g., $ \sigma_0 < \infty $ in normal-normal models), as the prior information can introduce a bias from the frequentist perspective. Understanding the frequentist bias and coverage probability in these cases can provide insight into how much the prior affects the results and whether the Bayesian estimates are suitable as frequentist proxies.


### Exercise 1: Frequentist characteristics of estimators in the normal-normal model

Consider the normal-normal model:

$$
(y_1, \dots, y_n | \mu) \sim \text{iid} \ N(\mu, \sigma^2) \quad \text{and} \quad \mu \sim N(\mu_0, \sigma_0^2)
$$

We are tasked with finding the general formulae for the **frequentist bias**, **relative bias**, and **frequentist coverage probability** of the posterior mean of $ \mu $, and producing corresponding graphs.

#### Solution:

Recall that $ (\mu | y) \sim N(\mu^*, \sigma^{*2}) $, where:

- $ \mu^* = (1 - k) \mu_0 + k \bar{y} $ is the posterior mean of $ \mu $
- $ \sigma^{*2} = \frac{\sigma^2}{n} k $ is the posterior variance of $ \mu $
- $ k = \frac{n \sigma_0^2}{n \sigma_0^2 + \sigma^2} $ is the credibility factor.

Also, the highest posterior density region (HPDR) for \( \mu \) is \( (\mu^* \pm z_{\alpha / 2} \sigma^*) \).

### 1. Frequentist Bias of the Posterior Mean

The frequentist bias of the posterior mean \( \mu^* \) is given by:

$$
B_{\mu} = E[\mu^* - \mu | \mu]
$$

Substitute \( \mu^* = (1 - k) \mu_0 + k \mu \):

$$
B_{\mu} = E[(1 - k) \mu_0 + k \mu - \mu | \mu] = (1 - k)(\mu_0 - \mu)
$$

Thus, the frequentist bias is:

$$
B_{\mu} = (1 - k)(\mu_0 - \mu)
$$

### 2. Frequentist Relative Bias

The **frequentist relative bias** of the posterior mean is defined as:

$$
R_{\mu} = \frac{B_{\mu}}{\mu} = \frac{(1 - k)(\mu_0 - \mu)}{\mu}
$$

Rewriting:

$$
R_{\mu} = (1 - k) \left( \frac{\mu_0}{\mu} - 1 \right), \quad (\mu \neq 0)
$$

### 3. Frequentist Coverage Probability of the HPDR

The **frequentist coverage probability** of the HPDR for \( \mu \) is given by:

$$
C_{\mu} = P\left( \mu \in (\mu^* \pm z_{\alpha / 2} \sigma^*) \mid \mu \right)
$$

Substitute \( \mu^* = (1 - k) \mu_0 + k \bar{y} \) and \( \sigma^* = \frac{\sigma}{\sqrt{n}} k \):

$$
C_{\mu} = P\left( \mu^* - z_{\alpha / 2} \sigma^* < \mu < \mu^* + z_{\alpha / 2} \sigma^* \mid \mu \right)
$$

Rearrange to get bounds for \( \bar{y} \):

$$
P\left( \frac{\mu - (1 - k)\mu_0 + z_{\alpha / 2} \sigma^*}{k} < \bar{y} < \frac{\mu - (1 - k)\mu_0 - z_{\alpha / 2} \sigma^*}{k} \mid \mu \right)
$$

This simplifies to:

$$
C_{\mu} = P\left( a(\mu) < \bar{y} < b(\mu) \mid \mu \right)
$$

where:

$$
a(\mu) = \frac{\mu - (1 - k) \mu_0 - z_{\alpha / 2} \sigma^*}{k}, \quad b(\mu) = \frac{\mu - (1 - k) \mu_0 + z_{\alpha / 2} \sigma^*}{k}
$$

Finally, the frequentist coverage probability becomes:

$$
C_{\mu} = \Phi\left( \frac{b(\mu) - \mu}{\sigma / \sqrt{n}} \right) - \Phi\left( \frac{a(\mu) - \mu}{\sigma / \sqrt{n}} \right)
$$

where $ \Phi $ is the standard normal cumulative distribution function (CDF).

### Note:
The functions $ a(\mu) $ and $ b(\mu) $ define the bounds of the credible interval for the parameter $ \mu $, and the frequentist coverage is derived by calculating the probability that the interval contains the true $ \mu $.

Graphs for each of these functions can be generated based on specific values of $ \mu_0 $, $ \sigma_0^2 $, $ \sigma^2 $, and $ n $, using numerical methods or simulations.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the normal-normal model
mu_0 = 0  # Prior mean
sigma_0 = 2  # Prior standard deviation
sigma = 1  # Likelihood standard deviation
alpha = 0.05  # Confidence level (for HPDR)
n = 30  # Sample size

# Credibility factor
def credibility_factor(n, sigma_0, sigma):
    return n * sigma_0**2 / (n * sigma_0**2 + sigma**2)

# Posterior mean
def posterior_mean(mu_0, y_bar, k):
    return (1 - k) * mu_0 + k * y_bar

# Posterior standard deviation
def posterior_std(sigma, n, k):
    return (sigma / np.sqrt(n)) * k

# Frequentist bias
def frequentist_bias(mu_0, mu, k):
    return (1 - k) * (mu_0 - mu)

# Frequentist relative bias
def relative_bias(mu_0, mu, k):
    return frequentist_bias(mu_0, mu, k) / mu if mu != 0 else np.nan

# Frequentist coverage probability for HPDR
def coverage_probability(mu_0, mu, sigma, n, alpha, k):
    z_alpha = norm.ppf(1 - alpha / 2)
    sigma_star = posterior_std(sigma, n, k)
    
    a_mu = (mu - (1 - k) * mu_0 - z_alpha * sigma_star) / k
    b_mu = (mu - (1 - k) * mu_0 + z_alpha * sigma_star) / k
    
    # Coverage probability
    coverage = norm.cdf(b_mu - mu, scale=sigma/np.sqrt(n)) - norm.cdf(a_mu - mu, scale=sigma/np.sqrt(n))
    
    return coverage

# Set up a range of mu values
mu_values = np.linspace(-5, 5, 100)

# Compute frequentist bias, relative bias, and coverage probability
k = credibility_factor(n, sigma_0, sigma)
bias_values = [frequentist_bias(mu_0, mu, k) for mu in mu_values]
rel_bias_values = [relative_bias(mu_0, mu, k) for mu in mu_values]
coverage_values = [coverage_probability(mu_0, mu, sigma, n, alpha, k) for mu in mu_values]

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(8, 12))

# Frequentist bias plot
ax[0].plot(mu_values, bias_values, label='Frequentist Bias')
ax[0].set_title('Frequentist Bias of the Posterior Mean')
ax[0].set_xlabel('$\mu$')
ax[0].set_ylabel('$B_{\mu}$')
ax[0].grid(True)

# Relative bias plot
ax[1].plot(mu_values, rel_bias_values, label='Relative Bias', color='orange')
ax[1].set_title('Frequentist Relative Bias of the Posterior Mean')
ax[1].set_xlabel('$\mu$')
ax[1].set_ylabel('$R_{\mu}$')
ax[1].grid(True)

# Coverage probability plot
ax[2].plot(mu_values, coverage_values, label='Coverage Probability', color='green')
ax[2].set_title('Frequentist Coverage Probability of HPDR for $\mu$')
ax[2].set_xlabel('$\mu$')
ax[2].set_ylabel('$C_{\mu}$')
ax[2].axhline(1 - alpha, color='red', linestyle='--', label='Nominal Coverage $(1 - \\alpha)$')
ax[2].legend()
ax[2].grid(True)

plt.tight_layout()
plt.show()
```

## Exercise 2: Frequentist Characteristics of Estimators in the Normal-Gamma Model

### Model Setup

We are given the following normal-gamma model:

$$
(y_1, y_2, \dots, y_n \mid \lambda) \sim \text{iid} \, N(\mu, \frac{1}{\lambda})
$$

where:

$$
\lambda \sim \text{Gamma}(\alpha, \beta)
$$

Our goal is to compute the frequentist bias, relative bias, and coverage probability of the posterior mean of $ \sigma^2 = \frac{1}{\lambda} $, and to find suitable priors $ \alpha $ and $ \beta $ such that the posterior mean is unbiased in the frequentist sense, and the CPDR has the desired coverage probability.

---

### Part (a): General Formulae for Bias, Relative Bias, and Coverage Probability

1. **Posterior Mean of $ \sigma^2 $:**

   The posterior mean of $ \sigma^2 $ is given by:

   $$
   \hat{\sigma}^2 = E(\sigma^2 \mid y) = \frac{b}{a-1}
   $$

   where:

   $$
   a = \alpha + \frac{n}{2}
   $$

   and

   $$
   b = \beta + \frac{s_\mu^2}{2}, \quad s_\mu^2 = \sum_{i=1}^{n} (y_i - \mu)^2
   $$

   Therefore, the posterior mean becomes:

   $$
   \hat{\sigma}^2 = \frac{2\beta + n s_\mu^2}{2\alpha + n - 2}
   $$

2. **Frequentist Bias:**

   The frequentist bias of $ \hat{\sigma}^2 $ is given by:

   $$
   B_{\sigma^2} = E(\hat{\sigma}^2 - \sigma^2 \mid \sigma^2) = \frac{2\beta + n E(s_\mu^2 \mid \sigma^2)}{2\alpha + n - 2} - \sigma^2
   $$

   Since:

   $$
   E(s_\mu^2 \mid \sigma^2) = n\sigma^2
   $$

   The bias simplifies to:

   $$
   B_{\sigma^2} = \frac{2\beta + n\sigma^2}{2\alpha + n - 2} - \sigma^2
   $$

3. **Frequentist Relative Bias:**

   The relative bias is given by:

   $$
   R_{\sigma^2} = \frac{B_{\sigma^2}}{\sigma^2}
   $$

   Substituting the bias expression:

   $$
   R_{\sigma^2} = \frac{\frac{2\beta + n\sigma^2}{2\alpha + n - 2} - \sigma^2}{\sigma^2}
   $$

   Simplifying further:

   $$
   R_{\sigma^2} = \frac{(2\beta / \sigma^2) + n}{2\alpha + n - 2} - 1
   $$

---

### Part (b): Finding a Suitable Prior

Our goal is to find $ \alpha $ and $ \beta $ such that:

1. The Bayesian posterior mean of $ \sigma^2 $ is unbiased in the frequentist sense for all possible values of $ \sigma^2 $:

   $$
   E[\hat{\sigma}^2 \mid \sigma^2] = \sigma^2
   $$

2. The CPDR for $ \sigma^2 $ has frequentist coverage probabilities exactly equal to the desired coverage for all values of $ \sigma^2 $.

To achieve this, we would need to explore values of $ \alpha $ and $ \beta $ that satisfy these conditions.

---

### Conclusion

The solution outlines the general formulae for the frequentist bias, relative bias, and the posterior mean under the normal-gamma model, and provides insights on how to select the prior parameters $ \alpha $ and $ \beta $ to meet the desired frequentist properties.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, norm

# Parameters of the normal-gamma model
alpha = 2  # Shape parameter of Gamma prior
beta = 2   # Rate parameter of Gamma prior
n = 30     # Sample size
mu = 0     # Mean of the normal distribution
alpha_cred = 0.05  # Credible interval level

# Posterior mean of sigma^2 (variance)
def posterior_mean_sigma2(y, mu, n, alpha, beta):
    s_sq = np.var(y)  # Sample variance
    return (beta + 0.5 * n * s_sq) / (alpha + 0.5 * n - 1)

# Frequentist bias of the posterior mean of sigma^2
def frequentist_bias_sigma2(sigma2, alpha, beta, n):
    return (beta + 0.5 * n * sigma2) / (alpha + 0.5 * n - 1) - sigma2

# Relative bias of the posterior mean of sigma2
def relative_bias_sigma2(sigma2, alpha, beta, n):
    bias = frequentist_bias_sigma2(sigma2, alpha, beta, n)
    return bias / sigma2

# Coverage probability of CPDR for sigma2
def coverage_probability_sigma2(sigma2, alpha, beta, n, alpha_cred):
    z_alpha = norm.ppf(1 - alpha_cred / 2)
    s_sq = sigma2
    sigma_post = np.sqrt(sigma2 / n)
    a = (sigma2 - z_alpha * sigma_post) / sigma_post
    b = (sigma2 + z_alpha * sigma_post) / sigma_post
    return norm.cdf(b) - norm.cdf(a)

# Range of sigma^2 values to analyze
sigma2_values = np.linspace(0.1, 5, 100)

# Compute frequentist bias, relative bias, and coverage probability
bias_values = [frequentist_bias_sigma2(sigma2, alpha, beta, n) for sigma2 in sigma2_values]
rel_bias_values = [relative_bias_sigma2(sigma2, alpha, beta, n) for sigma2 in sigma2_values]
coverage_values = [coverage_probability_sigma2(sigma2, alpha, beta, n, alpha_cred) for sigma2 in sigma2_values]

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(8, 12))

# Frequentist bias plot
ax[0].plot(sigma2_values, bias_values, label='Frequentist Bias')
ax[0].set_title('Frequentist Bias of the Posterior Mean of $\\sigma^2$')
ax[0].set_xlabel('$\\sigma^2$')
ax[0].set_ylabel('$B_{\\sigma^2}$')
ax[0].grid(True)

# Relative bias plot
ax[1].plot(sigma2_values, rel_bias_values, label='Relative Bias', color='orange')
ax[1].set_title('Relative Bias of the Posterior Mean of $\\sigma^2$')
ax[1].set_xlabel('$\\sigma^2$')
ax[1].set_ylabel('$R_{\\sigma^2}$')
ax[1].grid(True)

# Coverage probability plot
ax[2].plot(sigma2_values, coverage_values, label='Coverage Probability', color='green')
ax[2].set_title('Frequentist Coverage Probability of CPDR for $\\sigma^2$')
ax[2].set_xlabel('$\\sigma^2$')
ax[2].set_ylabel('$C_{\\sigma^2}$')
ax[2].axhline(1 - alpha_cred, color='red', linestyle='--', label='Nominal Coverage $(1 - \\alpha)$')
ax[2].legend()
ax[2].grid(True)

plt.tight_layout()
plt.show()
```

This further simplifies to:

$$
C_{\sigma^2} = P\left( \frac{ns_\mu^2}{\sigma^2} < v - \frac{2\beta}{\sigma^2}, \, u - \frac{2\beta}{\sigma^2} < \frac{ns_\mu^2}{\sigma^2} \mid \sigma^2 \right)
$$

or equivalently:

$$
C_{\sigma^2} = F_{\chi^2}(n) \left( \frac{v - \frac{2\beta}{\sigma^2}}{\sigma^2} \right) - F_{\chi^2}(n) \left( \frac{u - \frac{2\beta}{\sigma^2}}{\sigma^2} \right)
$$

---

### Part (b)

We now analyze specific prior settings for $ \alpha $ and $ \beta $.

1. **Under the prior $ \alpha = 1 $ and $ \beta = 0 $:**

   - The posterior mean of $ \sigma^2 $ equals the MLE, namely $ s_\mu^2 $, and thus is unbiased:

     $$
     \hat{\sigma}^2 = s_\mu^2
     $$

   - The $ 1 - A $ CPDR for $ \sigma^2 $ is:

     $$
     \left( \frac{ns_\mu^2}{\chi^2_{A/2}(n + 2)}, \frac{ns_\mu^2}{\chi^2_{1 - A/2}(n + 2)} \right)
     $$

     which has a coverage probability less than $ 1 - A $ for all $ \sigma^2 $.

2. **Under the prior $ \alpha = \beta = 0 $ (i.e., $ f(\lambda) \propto \lambda^{-1} $):**

   - The posterior mean of $ \sigma^2 $ is biased:

     $$
     \hat{\sigma}^2 = \frac{s_\mu^2}{1 - \frac{2}{n}}
     $$

   - The $ 1 - A $ CPDR for $ \sigma^2 $ is the same as the classical confidence interval (CI):

     $$
     \left( \frac{ns_\mu^2}{\chi^2_{A/2}(n)}, \frac{ns_\mu^2}{\chi^2_{1 - A/2}(n)} \right)
     $$

     and so has **exactly** $ 1 - A $ coverage probability for all $ \sigma^2 $.

---

### Conclusion

We conclude that there is no single gamma prior for $\lambda = \frac{1}{\sigma^2} $ that results in both:

1. A Bayesian posterior mean of $ \sigma^2 $ that is unbiased (in the frequentist sense) for all possible values of $ \sigma^2 $; and
2. A CPDR for $ \sigma^2 $ that has frequentist coverage probabilities exactly equal to the desired coverage for all possible values of $ \sigma^2 $.

---

### Correcting the Posterior Mean

It is easy to correct the posterior mean under $ \alpha = \beta = 0 $ so that it becomes unbiased. Explicitly, if $ \alpha = \beta = 0 $, then:

$$
E(\hat{\sigma}^2 \mid \sigma^2) = \frac{n \sigma^2}{n - 2}
$$

So an unbiased estimate of \( \sigma^2 \) is:

$$
\tilde{\sigma}^2 = \hat{\sigma}^2 \times \frac{n - 2}{n} = s_\mu^2
$$

which is, in fact, the MLE.
