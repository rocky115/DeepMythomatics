## Gamma and Weibull Distributions

The **gamma distribution** with parameters $\alpha$ and $n$ has the PDF:

$$f(x) = \frac{\alpha^n x^{n-1} e^{-\alpha x}}{\Gamma(n)}, \quad x > 0$$

where $\Gamma(n) = (n-1)!$ for integer $n$.

The mean of the gamma distribution is:

$$E(X) = \frac{n}{\alpha}$$

and the variance is $\frac{n}{\alpha^2}$.

Note that the mean and variance are $n$ times the mean and variance of the exponential distribution with parameter $\alpha$.

If $X_1, X_2, ..., X_n$ are independent and identically distributed exponential random variables, then $Y = \sum_{i=1}^{n} X_i$ has a gamma distribution with parameters $\alpha$ and $n$.

The **Weibull distribution** with parameters $\alpha$ and $\beta$ has the PDF:

$$f(x) = \alpha \beta x^{\beta - 1} e^{-\alpha x^\beta}, \quad x > 0$$

This distribution is used to model more complex lifetime data, especially when a constant failure rate is not a reasonable assumption.

Note that setting $\beta = 1$ gives the exponential distribution.

The mean and variance of the Weibull distribution are:

$$\mu = \frac{1}{\alpha^{1/\beta}} \Gamma\left(1 + \frac{1}{\beta}\right)$$

$$\sigma^2 = \frac{1}{\alpha^{2/\beta}} \left[\Gamma\left(1 + \frac{2}{\beta}\right) - \left(\Gamma\left(1 + \frac{1}{\beta}\right)\right)^2\right]$$

## 1.9 Generating Functions

We will consider two generating functions: the moment generating function (MGF) and the probability generating function (PGF). They are used to identify the probability distributions of functions of random variables.

The **moment generating function (MGF)** $M_X(s)$ is defined as:

$$M_X(s) = E(e^{sX})$$

It uses the series expansion of $e^{sX}$ to generate the moments $E(X^r)$ for $r \geq 1$.

The $r$-th moment of the random variable $X$ about zero is $E(X^r)$.

The power series expansion of $e^{sX}$ is:

$$e^{sX} = 1 + sX + \frac{(sX)^2}{2!} + ... + \frac{(sX)^r}{r!} + ...$$

which converges for all $sX$.

The MGF is obtained by taking expected values of both sides:

$$M_X(s) = E(e^{sX}) = E\left[1 + sX + \frac{(sX)^2}{2!} + ... + \frac{(sX)^r}{r!} + ...\right]$$

```python
import math

def gamma_pdf(x, alpha, n):
    """Calculates the PDF for a gamma distribution."""
    if x <= 0:
        return 0
    return (alpha**n * x**(n - 1) * math.exp(-alpha * x)) / math.gamma(n)

def gamma_mean(alpha, n):
    """Calculates the mean of a gamma distribution."""
    return n / alpha

def gamma_variance(alpha, n):
    """Calculates the variance of a gamma distribution."""
    return n / (alpha**2)

def weibull_pdf(x, alpha, beta):
    """Calculates the PDF for a Weibull distribution."""
    if x <= 0:
        return 0
    return alpha * beta * (x**(beta - 1)) * math.exp(-alpha * (x**beta))

def weibull_mean(alpha, beta):
    """Calculates the mean of a Weibull distribution."""
    return (1 / (alpha**(1 / beta))) * math.gamma(1 + (1 / beta))

def weibull_variance(alpha, beta):
    """Calculates the variance of a Weibull distribution."""
    term1 = math.gamma(1 + (2 / beta))
    term2 = (math.gamma(1 + (1 / beta)))**2
    return (1 / (alpha**(2 / beta))) * (term1 - term2)

def moment_generating_function_discrete(values, probabilities, s):
    """Calculates the MGF for a discrete random variable."""
    mgf = 0
    for x, p in zip(values, probabilities):
        mgf += math.exp(s * x) * p
    return mgf

def moment_generating_function_continuous(pdf, lower_bound, upper_bound, s, num_samples=1000):
    """Approximates the MGF for a continuous random variable."""
    delta = (upper_bound - lower_bound) / num_samples
    mgf = 0
    for i in range(num_samples):
        x = lower_bound + i * delta + delta / 2
        mgf += math.exp(s * x) * pdf(x) * delta
    return mgf

# Example Usage:

# Gamma Distribution:
alpha_gamma = 0.5
n_gamma = 2
x_gamma = 3
print(f"Gamma PDF (x={x_gamma}): {gamma_pdf(x_gamma, alpha_gamma, n_gamma)}")
print(f"Gamma Mean: {gamma_mean(alpha_gamma, n_gamma)}")
print(f"Gamma Variance: {gamma_variance(alpha_gamma, n_gamma)}")

# Weibull Distribution:
alpha_weibull = 0.5
beta_weibull = 2
x_weibull = 3
print(f"Weibull PDF (x={x_weibull}): {weibull_pdf(x_weibull, alpha_weibull, beta_weibull)}")
print(f"Weibull Mean: {weibull_mean(alpha_weibull, beta_weibull)}")
print(f"Weibull Variance: {weibull_variance(alpha_weibull, beta_weibull)}")

# Moment Generating Function (Discrete):
discrete_values = [0, 1, 2]
discrete_probabilities = [0.2, 0.5, 0.3]
s_discrete = 0.1
print(f"MGF (Discrete): {moment_generating_function_discrete(discrete_values, discrete_probabilities, s_discrete)}")

# Moment Generating Function (Continuous, Exponential):
alpha_exp = 1.0
s_continuous = 0.1
exp_pdf = lambda x: math.exp(-alpha_exp * x) if x >= 0 else 0
print(f"MGF (Continuous, Exponential): {moment_generating_function_continuous(exp_pdf, 0, 10, s_continuous)}")
```

##  Generating Functions (Continued)

If $X$ is an exponential random variable, then its MGF is:

$$M_X(s) = \frac{\alpha}{\alpha - s}$$

And if $Y = \sum_{i=1}^n X_i$, where $X_i$ are independent and identically distributed (iid) exponential random variables, then:

$$M_Y(s) = \left(\frac{\alpha}{\alpha - s}\right)^n$$

which is the MGF of a gamma-distributed random variable.

**Example 1.7:**

$n$ rails are cut to a nominal length of $\ell$ meters. The cutting machine cannot cut lengths less than $\ell$, but the actual length of any rail can have a positive error $x_i$, which is exponentially distributed with parameter $\alpha$ independently of each other. The rails are welded end-to-end to form a continuous rail of nominal length $n\ell$. Find the expected value and variance of the length of the composite rail.

**Solution:**

Let $X_i$ be the random variable representing the error $x_i$ of rail $i$. Let $Y$ be the random variable of the additional length of the whole rail, so that $Y = \sum_{i=1}^n X_i$.

Since the random variables are iid and exponentially distributed, the MGF is:

$$M_Y(s) = \prod_{i=1}^n M_{X_i}(s) = [M_{X_1}(s)]^n = \left(\frac{\alpha}{\alpha - s}\right)^n$$

Expanding this:

$$M_Y(s) = \left(1 - \frac{s}{\alpha}\right)^{-n} = 1 + \frac{n}{\alpha}s + \frac{n(n+1)}{2\alpha^2}s^2 + ...$$

The expected value of the length of the composite rail and the variance of the error are:

$$E(Y) = \frac{n}{\alpha}, \quad \sigma^2 = V(Y) = E(Y^2) - [E(Y)]^2 = \frac{n(n+1)}{\alpha^2} - \frac{n^2}{\alpha^2} = \frac{n}{\alpha^2}$$

This result shows that $V(\sum_{i=1}^n X_i) = \sum_{i=1}^n V(X_i)$ if $X_1, X_2, ..., X_n$ are independent.

The moment generating function (MGF) of a $N(\mu, \sigma^2)$ variable is:

$$M_X(s) = \exp\left(\mu s + \frac{1}{2}\sigma^2 s^2\right)$$

The MGF of a standard normal $Z$ is:

$$M_Z(s) = \exp\left(\frac{1}{2}s^2\right)$$

Consider a sequence of $n$ iid $N(\mu, \sigma^2)$ random variables $X_1, X_2, ..., X_n$. Then for the random variable $Y = \sum_{i=1}^n X_i$, using the results above:

$$M_Y(s) = \left[\exp\left(\mu s + \frac{1}{2}\sigma^2 s^2\right)\right]^n = \exp\left(n\mu s + \frac{1}{2}n\sigma^2 s^2\right)$$

```python
import math

def gamma_mgf(s, alpha, n):
    """Calculates the MGF for a gamma distribution."""
    return (alpha / (alpha - s)) ** n

def exponential_mgf(s, alpha):
    """Calculates the MGF for an exponential distribution."""
    return alpha / (alpha - s)

def normal_mgf(s, mu, sigma):
    """Calculates the MGF for a normal distribution."""
    return math.exp(mu * s + 0.5 * (sigma**2) * (s**2))

def standard_normal_mgf(s):
    """Calculates the MGF for a standard normal distribution."""
    return math.exp(0.5 * (s**2))

def sum_iid_normal_mgf(s, mu, sigma, n):
    """Calculates the MGF for the sum of n iid normal variables."""
    return math.exp(n * mu * s + 0.5 * n * (sigma**2) * (s**2))

def example_1_7(alpha, n):
    """Calculates the expected value and variance for Example 1.7."""
    expected_value = n / alpha
    variance = n / (alpha**2)
    return expected_value, variance

# Example Usage:

# Gamma MGF:
alpha_gamma = 2
n_gamma = 3
s_gamma = 0.5
print(f"Gamma MGF: {gamma_mgf(s_gamma, alpha_gamma, n_gamma)}")

# Exponential MGF:
alpha_exp = 1.5
s_exp = 0.2
print(f"Exponential MGF: {exponential_mgf(s_exp, alpha_exp)}")

# Normal MGF:
mu_normal = 5
sigma_normal = 2
s_normal = 0.3
print(f"Normal MGF: {normal_mgf(s_normal, mu_normal, sigma_normal)}")

# Standard Normal MGF:
s_std_normal = 0.4
print(f"Standard Normal MGF: {standard_normal_mgf(s_std_normal)}")

# Sum of IID Normal MGF:
mu_sum_normal = 1
sigma_sum_normal = 2
n_sum_normal = 3
s_sum_normal = 0.2
print(f"Sum IID Normal MGF: {sum_iid_normal_mgf(s_sum_normal, mu_sum_normal, sigma_sum_normal, n_sum_normal)}")

# Example 1.7:
alpha_example_1_7 = 0.1
n_example_1_7 = 5
expected, variance = example_1_7(alpha_example_1_7, n_example_1_7)
print(f"Example 1.7 Expected Value: {expected}")
print(f"Example 1.7 Variance: {variance}")
```

## 1.9 Generating Functions (Continued)

If the sequence of independent normally distributed random variables $X_1, X_2, ..., X_n$ have different means $\mu_i$ and variances $\sigma_i^2$, then $Y = \sum_{i=1}^n X_i$ is normally distributed with mean $\sum_{i=1}^n \mu_i$ and variance $\sum_{i=1}^n \sigma_i^2$.

A further generalization is to consider a linear sum of these random variables $W = \sum_{i=1}^n a_i X_i$, where the $a_i$'s are constants. Then $W$ is normally distributed with mean $\sum_{i=1}^n a_i \mu_i$ and variance $\sum_{i=1}^n a_i^2 \sigma_i^2$.

The **probability generating function (PGF)** $G_N(s)$ is defined for integer-valued random variables $N$ that can take values $0, 1, 2, ...$:

$$G_N(s) = E(s^N) = \sum_{n=0}^{\infty} p_n s^n$$

where $p_n = P(N = n)$.

The PGF is a power series in the dummy variable $s$.

Two random variables $X$ and $Y$ have the same PGF if and only if they have the same probability distributions.

If the PGF is a polynomial, it indicates a finite number of outcomes. If it is an infinite power series, it indicates a potentially unbounded number of outcomes.

If $G_N(s) = \sum_{n=0}^{\infty} p_n s^n$, then:

* First derivative: $G_N'(s) = \sum_{n=1}^{\infty} n p_n s^{n-1}$
* Second derivative: $G_N''(s) = \sum_{n=2}^{\infty} n(n-1) p_n s^{n-2}$

The PGF has the following properties:

* (a) $G_N(1) = \sum_{n=0}^{\infty} p_n = 1$
* (b) $G_N'(1) = \sum_{n=0}^{\infty} n p_n = E(N) = \mu$ (mean)

```python
import math

def normal_mgf(s, mu, sigma):
    """Calculates the MGF for a normal distribution."""
    return math.exp(mu * s + 0.5 * (sigma**2) * (s**2))

def sum_iid_normal_mgf(s, mu, sigma, n):
    """Calculates the MGF for the sum of n iid normal variables."""
    return math.exp(n * mu * s + 0.5 * n * (sigma**2) * (s**2))

def probability_generating_function(probabilities, s):
    """Calculates the probability generating function (PGF)."""
    pgf = 0
    for n, p_n in enumerate(probabilities):
        pgf += p_n * (s**n)
    return pgf

def pgf_derivative(probabilities, s):
    """Calculates the first derivative of the PGF."""
    derivative = 0
    for n, p_n in enumerate(probabilities):
        if n > 0:
            derivative += n * p_n * (s**(n - 1))
    return derivative

def pgf_second_derivative(probabilities, s):
    """Calculates the second derivative of the PGF."""
    second_derivative = 0
    for n, p_n in enumerate(probabilities):
        if n > 1:
            second_derivative += n * (n - 1) * p_n * (s**(n - 2))
    return second_derivative

def pgf_mean(probabilities):
    """Calculates the mean from the PGF."""
    return pgf_derivative(probabilities, 1)

# Example Usage:

# Normal MGF:
mu_normal = 5
sigma_normal = 2
s_normal = 0.3
print(f"Normal MGF: {normal_mgf(s_normal, mu_normal, sigma_normal)}")

# Sum of IID Normal MGF:
mu_sum_normal = 1
sigma_sum_normal = 2
n_sum_normal = 3
s_sum_normal = 0.2
print(f"Sum IID Normal MGF: {sum_iid_normal_mgf(s_sum_normal, mu_sum_normal, sigma_sum_normal, n_sum_normal)}")

# Probability Generating Function:
probabilities = [0.1, 0.2, 0.3, 0.4]  # Example probabilities p0, p1, p2, p3
s_pgf = 0.5
print(f"PGF: {probability_generating_function(probabilities, s_pgf)}")

# Derivatives of PGF:
print(f"PGF Derivative: {pgf_derivative(probabilities, s_pgf)}")
print(f"PGF Second Derivative: {pgf_second_derivative(probabilities, s_pgf)}")

# Mean from PGF:
print(f"PGF Mean: {pgf_mean(probabilities)}")
```

## 1.9 Generating Functions (Continued)

The PGF has the following additional properties:

* **(c)** $G_N''(1) = \sum_{n=0}^{\infty} n(n-1)p_n = E(N^2) - E(N)$. Therefore, the variance is:
    $$V(N) = \sigma^2 = E(N^2) - \mu^2 = G_N''(1) + G_N'(1) - [G_N'(1)]^2$$
* **(d)** $G_N^{(m)}(1) = \frac{d^m G(1)}{ds^m} = E[N(N-1)...(N-m+1)]$, which is the $m$-th factorial moment of $N$.
* **(e)** If $N_1, N_2, ..., N_r$ are independent and identically distributed (iid) discrete random variables, and $Y = \sum_{i=1}^r N_i$, then:
    $$G_Y(s) = E[s^Y] = E\left[s^{\sum_{i=1}^r N_i}\right] = E\left[\prod_{i=1}^r s^{N_i}\right] = \prod_{i=1}^r E[s^{N_i}] = \prod_{i=1}^r G_{N_i}(s)$$
    Since the $N_i$'s are identically distributed:
    $$G_Y(s) = [G_X(s)]^r$$

**Example 1.8:**

The random variable $N$ has a binomial distribution with parameters $m$ and $p$. Its probability function is:

$$p(n) = p_n = P(N = n) = \binom{m}{n} p^n q^{m-n}, \quad n = 0, 1, 2, ..., m$$

Find its PGF, mean, and variance.

**Solution:**

The PGF of $N$ is:

$$G_N(s) = G(s) = \sum_{n=0}^m s^n \binom{m}{n} p^n q^{m-n} = \sum_{n=0}^m \binom{m}{n} (ps)^n q^{m-n} = (q + ps)^m$$

using the binomial theorem.

The derivatives are:

$$G'(s) = mp(q + ps)^{m-1}$$
$$G''(s) = m(m-1)p^2(q + ps)^{m-2}$$

Using the properties above, the mean and variance are:

$$\mu = G'(1) = mp$$
$$\sigma^2 = G''(1) + G'(1) - [G'(1)]^2 = m(m-1)p^2 + mp - m^2p^2 = mpq$$

The Bernoulli distribution is the binomial distribution with $m = 1$. Hence, its PGF is:

$$G_X(s) = q + ps$$

```python
import math

def binomial_pgf(s, m, p):
    """Calculates the PGF for a binomial distribution."""
    q = 1 - p
    return (q + p * s) ** m

def binomial_mean_from_pgf(m, p):
    """Calculates the mean of a binomial distribution using its PGF."""
    return m * p

def binomial_variance_from_pgf(m, p):
    """Calculates the variance of a binomial distribution using its PGF."""
    q = 1 - p
    return m * p * q

def bernoulli_pgf(s, p):
    """Calculates the PGF for a Bernoulli distribution."""
    q = 1 - p
    return q + p * s

# Example Usage:

# Binomial Distribution:
m_binomial = 10
p_binomial = 0.3
s_binomial = 0.5
print(f"Binomial PGF: {binomial_pgf(s_binomial, m_binomial, p_binomial)}")
print(f"Binomial Mean (from PGF): {binomial_mean_from_pgf(m_binomial, p_binomial)}")
print(f"Binomial Variance (from PGF): {binomial_variance_from_pgf(m_binomial, p_binomial)}")

# Bernoulli Distribution:
p_bernoulli = 0.7
s_bernoulli = 0.8
print(f"Bernoulli PGF: {bernoulli_pgf(s_bernoulli, p_bernoulli)}")
```

Consider $n$ independent and identically distributed Bernoulli random variables $X_1, X_2, \dots, X_n$, and let $Y = \sum_{i=1}^n X_i$. Then from the results above, $G_Y(s) = [G_X(s)]^n = (q + ps)^n$, which is again the pgf of a binomial random variable. It is possible to associate a generating function with any sequence $\{a_n\}$, $(n = 0, 1, 2, \dots)$ in the form $$ H(s) = \sum_{n=0}^\infty a_n s^n $$ provided that the series converges in some interval containing the origin $s = 0$. Unlike the pgf, this series need not satisfy the conditions $H(1) = 1$ for a probability distribution nor $0 \leq a_n \leq 1$. An application using such a series is given in Problem 1.24, and in Section 3.3 on random walks.

### 1.10 Conditional Expectation

In many applications of probability we are interested in the possible values of two or more characteristics in a problem. For this we require two or more random variables which may or may not be independent. We shall only consider the case of two discrete random variables $X$ and $Y$ which form a two-dimensional random variable denoted by $(X, Y)$, which can take pairs of values $(x_i, y_j)$: for example, we could assume $(i = 1, 2, \dots ; j = 1, 2, \dots)$ (either sequence may be finite or infinite, or start for some other values) with joint probability (mass) function $p(x_i, y_j)$, which is now a function of two variables. As in Section 1.4, the probabilities must satisfy:

- (a) $0 \leq p(x_i, y_j) \leq 1$ for all $(i, j)$,
- (b) $$ \sum_{i \in I} \sum_{j \in J} p(x_i, y_j) = 1 $$

The domains of $i$ and $j$ are defined by $i \in I$ and $j \in J$ where $I$ and $J$ can be either bounded or unbounded sequences of the integers. The random variables $X$ and $Y$ are said to be independent, if and only if, $$ p(x_i, y_j) = q(x_i) r(y_j) $$ for all $i$ and $j$, where, of course, $$ \sum_{i \in I} q(x_i) = 1, \quad \sum_{j \in J} r(y_j) = 1 $$

If the random variable $Z = H(X, Y)$ is a function of the random variables $X$ and $Y$, then we will state without proof that the expected value of $Z$ is given by $$ E(Z) = \sum_{i \in I} \sum_{j \in J} H(x_i, y_j) p(x_i, y_j) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of Bernoulli trials
p = 0.4  # Success probability
q = 1 - p
num_simulations = 10000

# 1. Simulate sum of Bernoulli random variables
trials = np.random.binomial(n, p, num_simulations)
unique, counts = np.unique(trials, return_counts=True)
empirical_probs = counts / num_simulations

# Compute theoretical pgf: G_Y(s) = (q + ps)^n
s_values = np.linspace(0, 2, 100)
pgf_theoretical = (q + p * s_values) ** n

# Empirical pgf: E[s^Y] = sum P(Y=k) * s^k
pgf_empirical = np.zeros_like(s_values)
for k, prob in zip(unique, empirical_probs):
    pgf_empirical += prob * s_values ** k

# Plot pgf
plt.plot(s_values, pgf_theoretical, label='Theoretical pgf $(q + ps)^n$', color='blue')
plt.plot(s_values, pgf_empirical, '--', label='Empirical pgf', color='red')
plt.xlabel('$s$')
plt.ylabel('$G_Y(s)$')
plt.title('Probability Generating Function for Binomial')
plt.legend()
plt.grid(True)
plt.show()

# 2. Conditional Expectation Example
# Joint distribution: X (Bernoulli, p=0.5), Y = X + W (W is Bernoulli, p=0.3)
p_x = 0.5
p_w = 0.3
x_values = [0, 1]
y_values = [0, 1, 2]  # Possible values of Y = X + W

# Simulate joint distribution
x_samples = np.random.binomial(1, p_x, num_simulations)
w_samples = np.random.binomial(1, p_w, num_simulations)
y_samples = x_samples + w_samples

# Compute conditional expectation E[Y | X]
for x in x_values:
    y_given_x = y_samples[x_samples == x]
    if len(y_given_x) > 0:
        e_y_given_x = np.mean(y_given_x)
        print(f"Empirical E[Y | X = {x}] = {e_y_given_x:.3f}")

# Theoretical E[Y | X]
# If X = 0, Y = W, so E[Y | X=0] = E[W] = p_w
# If X = 1, Y = 1 + W, so E[Y | X=1] = 1 + E[W] = 1 + p_w
print(f"Theoretical E[Y | X = 0] = {p_w:.3f}")
print(f"Theoretical E[Y | X = 1] = {1 + p_w:.3f}")
```

# Conditional Expectation and Law of Total Expectation

Returning to the dependent case, we consider the conditional expectations $E(X|Y)$ and $E(Y|X)$, which are random variables and thus have their own expectations. The law of total expectation states:

$$ E(E(X|Y)) = \sum_{j \in J} E(X|Y = y_j) p_Y(y_j) $$

This can be computed as:

$$ E(E(X|Y)) = \sum_{j \in J} \sum_{k \in I} E(X|Y = y_j) p(x_k, y_j) $$

$$ = \sum_{j \in J} \sum_{i \in I} x_i p_X(x_i | y_j) p(x_k, y_j) $$

$$ = \sum_{j \in J} \sum_{i \in I} x_i p(x_i, y_j) $$

$$ = E(X) $$

This result shows that the expected value of the conditional expectation of $X$ with respect to $Y$ equals the expected value of $X$. Similarly:

$$ E(E(Y|X)) = E(Y) $$

## Example 1.9: Conditional Expectation Calculation

Consider a random variable $(X, Y)$ taking values $(x_i, y_j)$ for $i = 1, 2, 3$ and $j = 1, 2, 3$. The joint probability mass function $p(x_i, y_j)$ is given in the following table:

| $p(x_i, y_j)$ | $y_1$ | $y_2$ | $y_3$ |
|---------------|-------|-------|-------|
| $x_1$         | 0.25  | 0     | 0.05  |
| $x_2$         | 0.05  | 0.10  | 0.15  |
| $x_3$         | 0.05  | 0.25  | 0.10  |

We aim to find the random variable $E(X|Y)$ and verify that $E(E(X|Y)) = E(X)$.

### Step 1: Verify Joint Distribution

The indices are $I = \{1, 2, 3\}$ and $J = \{1, 2, 3\}$. The joint probabilities sum to 1:

$$ \sum_{i=1}^3 \sum_{j=1}^3 p(x_i, y_j) = 0.25 + 0 + 0.05 + 0.05 + 0.10 + 0.15 + 0.05 + 0.25 + 0.10 = 1 $$

This confirms the table represents a valid joint probability distribution.

### Step 2: Marginal Probabilities for $Y$

Compute the marginal probability distribution $p_Y(y_j)$:

- For $y_1$:

$$ p_Y(y_1) = \sum_{i=1}^3 p(x_i, y_1) = p(x_1, y_1) + p(x_2, y_1) + p(x_3, y_1) = 0.25 + 0.05 + 0.05 = 0.35 $$

- For $y_2$:

$$ p_Y(y_2) = \sum_{i=1}^3 p(x_i, y_2) = p(x_1, y_2) + p(x_2, y_2) + p(x_3, y_2) = 0 + 0.10 + 0.25 = 0.35 $$

- For $y_3$:

$$ p_Y(y_3) = \sum_{i=1}^3 p(x_i, y_3) = p(x_1, y_3) + p(x_2, y_3) + p(x_3, y_3) = 0.05 + 0.15 + 0.10 = 0.30 $$

The marginal probabilities sum to:

$$ p_Y(y_1) + p_Y(y_2) + p_Y(y_3) = 0.35 + 0.35 + 0.30 = 1 $$

This verifies the marginal distribution for $Y$.

```python
# Joint probability table: p(x_i, y_j)
# Rows: x_1, x_2, x_3; Columns: y_1, y_2, y_3
joint_prob = [
    [0.25, 0.00, 0.05],  # x_1
    [0.05, 0.10, 0.15],  # x_2
    [0.05, 0.25, 0.10]   # x_3
]

# Values of X
x_values = [1, 2, 3]

# Step 1: Compute marginal probabilities p_Y(y_j)
p_y = [0.0, 0.0, 0.0]  # p_Y(y_1), p_Y(y_2), p_Y(y_3)
for j in range(3):
    for i in range(3):
        p_y[j] += joint_prob[i][j]

print("Marginal probabilities p_Y(y_j):")
for j in range(3):
    print(f"p_Y(y_{j+1}) = {p_y[j]:.2f}")

# Step 2: Compute conditional probabilities p(X=x_i | Y=y_j) and E(X | Y=y_j)
e_x_given_y = [0.0, 0.0, 0.0]  # E(X | Y=y_j)
for j in range(3):
    if p_y[j] > 0:  # Avoid division by zero
        for i in range(3):
            p_x_given_y = joint_prob[i][j] / p_y[j]
            e_x_given_y[j] += x_values[i] * p_x_given_y

print("\nConditional expectations E(X | Y=y_j):")
for j in range(3):
    print(f"E(X | Y=y_{j+1}) = {e_x_given_y[j]:.4f}")

# Step 3: Compute E(E(X|Y))
e_e_x_given_y = 0.0
for j in range(3):
    e_e_x_given_y += e_x_given_y[j] * p_y[j]

# Step 4: Compute E(X) directly
# First, compute marginal p_X(x_i)
p_x = [0.0, 0.0, 0.0]  # p_X(x_1), p_X(x_2), p_X(x_3)
for i in range(3):
    for j in range(3):
        p_x[i] += joint_prob[i][j]

# Compute E(X)
e_x = 0.0
for i in range(3):
    e_x += x_values[i] * p_x[i]

print("\nVerification:")
print(f"E(E(X|Y)) = {e_e_x_given_y:.4f}")
print(f"E(X) = {e_x:.4f}")
print(f"Difference |E(E(X|Y)) - E(X)| = {abs(e_e_x_given_y - e_x):.4f}")
```

# Conditional Expectation: Example 1.9 (Continued)

Given the random variable $(X, Y)$ with values $(x_i, y_j)$ for $i = 1, 2, 3$ and $j = 1, 2, 3$, where $x_1 = 1$, $x_2 = 2$, $x_3 = 3$, the joint probability mass function $p(x_i, y_j)$ is:

| $p(x_i, y_j)$ | $y_1$ | $y_2$ | $y_3$ |
|---------------|-------|-------|-------|
| $x_1$         | 0.25  | 0     | 0.05  |
| $x_2$         | 0.05  | 0.10  | 0.15  |
| $x_3$         | 0.05  | 0.25  | 0.10  |

## Marginal Probabilities

The marginal probabilities for $Y$ were computed as:

- $$ p_Y(y_1) = \sum_{i=1}^3 p(x_i, y_1) = 0.25 + 0.05 + 0.05 = 0.35 $$

- $$ p_Y(y_2) = \sum_{i=1}^3 p(x_i, y_2) = 0 + 0.10 + 0.25 = 0.35 $$

- $$ p_Y(y_3) = \sum_{i=1}^3 p(x_i, y_3) = 0.05 + 0.15 + 0.10 = 0.30 $$

For $X$, the marginal probabilities are:

- $$ p_X(x_1) = \sum_{j=1}^3 p(x_1, y_j) = 0.25 + 0 + 0.05 = 0.30 $$

- $$ p_X(x_2) = \sum_{j=1}^3 p(x_2, y_j) = 0.05 + 0.10 + 0.15 = 0.30 $$

- $$ p_X(x_3) = \sum_{j=1}^3 p(x_3, y_j) = 0.05 + 0.25 + 0.10 = 0.40 $$

## Conditional Probabilities

Using the conditional probability formula:

$$ p(X = x_i | Y = y_j) = \frac{p(x_i, y_j)}{p_Y(y_j)} $$

We compute:

- For $Y = y_1$:

  $$ p(X = x_1 | y_1) = \frac{p(x_1, y_1)}{p_Y(y_1)} = \frac{0.25}{0.35} = \frac{5}{7} $$

  $$ p(X = x_2 | y_1) = \frac{0.05}{0.35} = \frac{1}{7} $$

  $$ p(X = x_3 | y_1) = \frac{0.05}{0.35} = \frac{1}{7} $$

- For $Y = y_2$:

  $$ p(X = x_1 | y_2) = \frac{0}{0.35} = 0 $$

  $$ p(X = x_2 | y_2) = \frac{0.10}{0.35} = \frac{2}{7} $$

  $$ p(X = x_3 | y_2) = \frac{0.25}{0.35} = \frac{5}{7} $$

- For $Y = y_3$:

  $$ p(X = x_1 | y_3) = \frac{0.05}{0.30} = \frac{1}{6} $$

  $$ p(X = x_2 | y_3) = \frac{0.15}{0.30} = \frac{1}{2} $$

  $$ p(X = x_3 | y_3) = \frac{0.10}{0.30} = \frac{1}{3} $$

## Conditional Expectation $E(X|Y)$

The conditional expectation is:

$$ E(X | Y = y_j) = \sum_{i=1}^3 x_i p(X = x_i | Y = y_j) $$

- For $Y = y_1$:

  $$ E(X | Y = y_1) = 1 \cdot \frac{5}{7} + 2 \cdot \frac{1}{7} + 3 \cdot \frac{1}{7} = \frac{5 + 2 + 3}{7} = \frac{10}{7} $$

- For $Y = y_2$:

  $$ E(X | Y = y_2) = 1 \cdot 0 + 2 \cdot \frac{2}{7} + 3 \cdot \frac{5}{7} = \frac{4 + 15}{7} = \frac{19}{7} $$

- For $Y = y_3$:

  $$ E(X | Y = y_3) = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{2} + 3 \cdot \frac{1}{3} = \frac{1 + 6 + 6}{6} = \frac{13}{6} $$

Thus, the random variable $E(X|Y)$ takes values:

$$ \left\{ \frac{10}{7}, \frac{19}{7}, \frac{13}{6} \right\} $$

## Law of Total Expectation

The expectation of $E(X|Y)$, denoted $Z = E(X|Y)$, is:

$$ E(Z) = \sum_{j=1}^3 E(X | Y = y_j) p_Y(y_j) = \frac{10}{7} \cdot 0.35 + \frac{19}{7} \cdot 0.35 + \frac{13}{6} \cdot 0.30 = \frac{21}{10} $$

Directly, the expectation of $X$ is:

$$ E(X) = \sum_{i=1}^3 x_i p_X(x_i) = 1 \cdot 0.30 + 2 \cdot 0.30 + 3 \cdot 0.40 = \frac{21}{10} $$

This confirms:

$$ E(E(X|Y)) = E(X) $$

```python
# Core Python implementation for Example 1.9

# Joint probability table: p(x_i, y_j)
# Rows: x_1, x_2, x_3; Columns: y_1, y_2, y_3
joint_prob = [
    [0.25, 0.00, 0.05],  # x_1
    [0.05, 0.10, 0.15],  # x_2
    [0.05, 0.25, 0.10]   # x_3
]

# Values of X
x_values = [1, 2, 3]

# Step 1: Compute marginal probabilities
# p_Y(y_j)
p_y = [0.0] * 3
for j in range(3):
    for i in range(3):
        p_y[j] += joint_prob[i][j]

print("Marginal probabilities p_Y(y_j):")
for j in range(3):
    print(f"p_Y(y_{j+1}) = {p_y[j]:.2f}")

# p_X(x_i)
p_x = [0.0] * 3
for i in range(3):
    for j in range(3):
        p_x[i] += joint_prob[i][j]

print("\nMarginal probabilities p_X(x_i):")
for i in range(3):
    print(f"p_X(x_{i+1}) = {p_x[i]:.2f}")

# Step 2: Compute conditional probabilities p(X=x_i | Y=y_j)
conditional_prob = [[0.0] * 3 for _ in range(3)]  # [i][j] = p(X=x_i | Y=y_j)
for j in range(3):
    if p_y[j] > 0:
        for i in range(3):
            conditional_prob[i][j] = joint_prob[i][j] / p_y[j]

print("\nConditional probabilities p(X=x_i | Y=y_j):")
for j in range(3):
    print(f"Y = y_{j+1}:")
    for i in range(3):
        print(f"  p(X=x_{i+1} | Y=y_{j+1}) = {conditional_prob[i][j]:.4f}")

# Step 3: Compute E(X | Y=y_j)
e_x_given_y = [0.0] * 3
for j in range(3):
    if p_y[j] > 0:
        for i in range(3):
            e_x_given_y[j] += x_values[i] * conditional_prob[i][j]

print("\nConditional expectations E(X | Y=y_j):")
for j in range(3):
    print(f"E(X | Y=y_{j+1}) = {e_x_given_y[j]:.4f}")

# Step 4: Compute E(E(X|Y))
e_e_x_given_y = 0.0
for j in range(3):
    e_e_x_given_y += e_x_given_y[j] * p_y[j]

# Step 5: Compute E(X)
e_x = 0.0
for i in range(3):
    e_x += x_values[i] * p_x[i]

# Step 6: Verify
print("\nVerification:")
print(f"E(E(X|Y)) = {e_e_x_given_y:.4f}")
print(f"E(X) = {e_x:.4f}")
print(f"Difference |E(E(X|Y)) - E(X)| = {abs(e_e_x_given_y - e_x):.4f}")
```
