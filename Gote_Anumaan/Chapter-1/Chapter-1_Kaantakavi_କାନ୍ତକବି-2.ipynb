## Example 1.3: Independent Events

Let A and B be independent events with $P(A) = \frac{1}{4}$ and $P(B) = \frac{2}{3}$. Calculate the following probabilities:

(a) $P(A \cap B)$
(b) $P(A \cap B^c)$
(c) $P(A^c \cap B^c)$
(d) $P(A^c \cap B)$
(e) $P((A \cup B)^c)$

**Solution:**

(a) Since A and B are independent, $P(A \cap B) = P(A)P(B)$.
   $$P(A \cap B) = \frac{1}{4} \cdot \frac{2}{3} = \frac{2}{12} = \frac{1}{6}$$

(b) Since A and B are independent, A and $B^c$ are also independent.
   $$P(A \cap B^c) = P(A)P(B^c) = P(A)(1 - P(B)) = \frac{1}{4} \left(1 - \frac{2}{3}\right) = \frac{1}{4} \cdot \frac{1}{3} = \frac{1}{12}$$

(c) Since A and B are independent, $A^c$ and $B^c$ are also independent.
   $$P(A^c \cap B^c) = P(A^c)P(B^c) = (1 - P(A))(1 - P(B)) = \left(1 - \frac{1}{4}\right)\left(1 - \frac{2}{3}\right) = \frac{3}{4} \cdot \frac{1}{3} = \frac{1}{4}$$

(d) Since A and B are independent, $A^c$ and B are also independent.
   $$P(A^c \cap B) = P(A^c)P(B) = (1 - P(A))P(B) = \left(1 - \frac{1}{4}\right)\left(\frac{2}{3}\right) = \frac{3}{4} \cdot \frac{2}{3} = \frac{1}{2}$$

(e) Using the formula $P((A \cup B)^c) = 1 - P(A \cup B) = 1 - [P(A) + P(B) - P(A \cap B)]$:
   $$P((A \cup B)^c) = 1 - \left(\frac{1}{4} + \frac{2}{3} - \frac{1}{6}\right) = 1 - \left(\frac{3}{12} + \frac{8}{12} - \frac{2}{12}\right) = 1 - \frac{9}{12} = 1 - \frac{3}{4} = \frac{1}{4}$$

## Example 1.4: Conditional Probability

For three events A, B, and C, show that $P(A \cap B|C) = P(A|B \cap C)P(B|C)$, where $P(C) > 0$.

**Solution:**

Using the definition of conditional probability:
$$P(A \cap B|C) = \frac{P((A \cap B) \cap C)}{P(C)} = \frac{P(A \cap B \cap C)}{P(C)}$$

Also,
$$P(A|B \cap C) = \frac{P(A \cap (B \cap C))}{P(B \cap C)} = \frac{P(A \cap B \cap C)}{P(B \cap C)}$$
and
$$P(B|C) = \frac{P(B \cap C)}{P(C)}$$

Therefore,
$$P(A|B \cap C)P(B|C) = \frac{P(A \cap B \cap C)}{P(B \cap C)} \cdot \frac{P(B \cap C)}{P(C)} = \frac{P(A \cap B \cap C)}{P(C)} = P(A \cap B|C)$$

Hence, $P(A \cap B|C) = P(A|B \cap C)P(B|C)$.

If the first attempt fails to open the case, the traveler tries another code from the remaining pairs. What is the probability that the new code opens the case, or any subsequent attempt?

**Solution:**

At the first attempt, the probability of opening the case is:
$$P(X_1) = \frac{1}{81}$$

Now, let's consider the second attempt. The probability that the second attempt opens the case, given that the first attempt failed, is:
$$P(X_2) = P(X_2 \cap X_1^c)$$

Using the conditional probability formula:
$$P(X_2) = P(X_1^c)P(X_2 | X_1^c)$$

The probability that the first attempt failed is:
$$P(X_1^c) = 1 - P(X_1) = 1 - \frac{1}{81} = \frac{80}{81}$$

If the first attempt failed, there are 80 remaining codes. So the probability that the second attempt opens the case, given the first failed, is:
$$P(X_2 | X_1^c) = \frac{1}{80}$$

Therefore, the probability of the second attempt opening the case is:
$$P(X_2) = \frac{80}{81} \cdot \frac{1}{80} = \frac{1}{81}$$

In general, for the $n$-th attempt, the probability of opening the case is:
$$P(X_n) = \frac{1}{81}$$

Note that the conditional probability $P(X_2 | X_1^c) = \frac{1}{80}$.

## Law of Total Probability (Partition Theorem)

Suppose that $A_1, A_2, ..., A_k$ represents a partition of the sample space $S$ into $k$ mutually exclusive and exhaustive events. This means:

* $A_i \cap A_j = \emptyset$ for $i \neq j$ (mutually exclusive)
* $\bigcup_{i=1}^{k} A_i = S$ (exhaustive)

Let $B$ be another event associated with the same random experiment. Then:

$$B = \bigcup_{i=1}^{k} (B \cap A_i)$$

And since the events $B \cap A_i$ are mutually exclusive, we have:

$$P(B) = \sum_{i=1}^{k} P(B \cap A_i) = \sum_{i=1}^{k} P(A_i)P(B|A_i)$$

This result is known as the law of total probability or the partition theorem.

## 1.3 Law of Total Probability (Partition Theorem) - Formalization

The law of total probability provides a way to calculate the probability of an event $B$ by considering its conditional probabilities given a partition of the sample space $S$.

Let $A_1, A_2, ..., A_k$ be a partition of the sample space $S$. This means:

* $A_i \cap A_j = \emptyset$ for $i \neq j$ (mutually exclusive)
* $\bigcup_{i=1}^{k} A_i = S$ (exhaustive)

Then, for any event $B$:

$$P(B) = \sum_{i=1}^{k} P(B \cap A_i)$$

Since $P(B \cap A_i) = P(B|A_i)P(A_i)$, we can rewrite the equation as:

$$P(B) = \sum_{i=1}^{k} P(B|A_i)P(A_i)$$

This is the law of total probability.

## 1.4 Discrete Random Variables

In many applications, the outcomes of a random experiment are numerical. A **random variable** (usually denoted by $X, Y, Z$) is a numerical value associated with the outcome of a random experiment.

If $s$ is an element of the sample space $S$, then $X(s)$ is a real number associated with $s$.

We are interested in probabilities like $P(X = x_1)$ or $P(X \leq x_2)$.

A **discrete random variable** is a random variable whose sample space is finite or countably infinite (e.g., integers).

The set of possible values of $X$ is denoted by $V = \{x_i\}$, where $x_i$ are real numbers.

Examples of $V$:

* $V = \{0, \frac{1}{2}, 1, \frac{3}{2}, 2, \frac{5}{2}, 3, ...\}$
* $V = \{0, 1\}$
* $V = \{0, 1, 2, 3, ...\}$
* $V = \{..., -3, -2, -1, 0, 1, 2, 3, ...\}$

The probability $p(x_i) = P(X = x_i)$ is called the **probability mass function (PMF)**.

The pairs $\{x_i, p(x_i)\}$ define the probability distribution of the random variable $X$.

If $x_i = i$, then $p(i) = P(X = i)$.

```python
def law_of_total_probability(partitions, conditional_probabilities):
    """
    Calculates the total probability of an event using the law of total probability.

    Args:
        partitions: A list of probabilities of the partition events (A_i).
        conditional_probabilities: A list of conditional probabilities P(B|A_i).

    Returns:
        The total probability of event B.
    """

    if len(partitions) != len(conditional_probabilities):
        raise ValueError("Number of partitions and conditional probabilities must be the same.")

    total_probability = 0
    for i in range(len(partitions)):
        total_probability += partitions[i] * conditional_probabilities[i]

    return total_probability

def discrete_random_variable_pmf(values, probabilities):
    """
    Creates a probability mass function (PMF) for a discrete random variable.

    Args:
        values: A list of possible values of the random variable (x_i).
        probabilities: A list of corresponding probabilities (p(x_i)).

    Returns:
        A dictionary representing the PMF, where keys are values and values are probabilities.
    """

    if len(values) != len(probabilities):
        raise ValueError("Number of values and probabilities must be the same.")

    pmf = dict(zip(values, probabilities))
    return pmf

def calculate_probability_of_event(pmf, event_condition):
    """
    Calculates the probability of an event for a discrete random variable given its PMF.

    Args:
        pmf: A dictionary representing the PMF.
        event_condition: A function that defines the event condition (e.g., lambda x: x <= 2).

    Returns:
        The probability of the event.
    """

    probability = 0
    for value, prob in pmf.items():
        if event_condition(value):
            probability += prob
    return probability

# Example usage for law of total probability:
partitions = [0.3, 0.4, 0.3]
conditional_probabilities = [0.1, 0.5, 0.8]
total_prob = law_of_total_probability(partitions, conditional_probabilities)
print(f"Total probability: {total_prob}")

# Example usage for discrete random variable PMF:
values = [0, 1, 2, 3]
probabilities = [0.1, 0.2, 0.3, 0.4]
pmf = discrete_random_variable_pmf(values, probabilities)
print(f"PMF: {pmf}")

# Example usage for calculating event probability:
event_probability = calculate_probability_of_event(pmf, lambda x: x <= 2)
print(f"Probability of X <= 2: {event_probability}")

event_probability_equal_1 = calculate_probability_of_event(pmf, lambda x: x == 1)
print(f"Probability of X == 1: {event_probability_equal_1}")
```

## 1.4 Discrete Random Variables (Continued)

If $x_i = i$, which occurs frequently in applications, then $p(x_i) = p(i)$ is often replaced by $p_i$.

Since the $x$ values are mutually exclusive and exhaustive, it follows that for $p(x_0), p(x_1), ...$:

* (a) $0 \leq p(x_i) \leq 1$ for all $i$
* (b) $\sum_{i=0}^{\infty} p(x_i) = 1$, or in generic form $\sum_{x \in V} p(x) = 1$
* (c) $P(X \leq x_k) = \sum_{i=0}^{k} p(x_i)$, which is known as the **distribution function** for $k = 0, 1, 2, ...$

**Example 1.5:**

A fair die is rolled until the first 6 appears face up. Find the probability that the first 6 appears at the $n$-th throw.

Let the random variable $N$ be the number of throws until the first 6 appears face up. This is a discrete random variable with an infinite number of possible outcomes $\{1, 2, 3, ...\}$.

The probability of a 6 appearing for any throw is $\frac{1}{6}$, and the probability of any other number appearing is $\frac{5}{6}$.

Hence, the probability of $n-1$ numbers other than 6 appearing followed by a 6 is:

$$P(N = n) = \left(\frac{5}{6}\right)^{n-1} \left(\frac{1}{6}\right) = \frac{5^{n-1}}{6^n}$$

This is the probability mass function (PMF) for this random variable.

The distribution function is:

$$P(N \leq k) = \sum_{i=1}^{k} \left(\frac{5}{6}\right)^{i-1} \left(\frac{1}{6}\right) = \frac{1}{6} \sum_{i=1}^{k} \left(\frac{5}{6}\right)^{i-1}$$

Using the formula for the sum of a geometric series:

$$P(N \leq k) = \frac{1}{6} \cdot \frac{1 - \left(\frac{5}{6}\right)^k}{1 - \frac{5}{6}} = 1 - \left(\frac{5}{6}\right)^k$$

for $k = 1, 2, ...$

## 1.5 Continuous Random Variables

In many applications, discrete random variables are not suitable for problems where the random variable can take any real value in an interval. For example, the time until a light bulb fails ($T$) can be any value $t \geq 0$.

In this case, $T$ is called a **continuous random variable**.

For continuous random variables, defining the event $X = x$ is mathematically difficult, and the probability is usually defined to be zero. Probabilities are defined over intervals, such as $P(x_1 < X < x_2)$.

We define a **probability density function (PDF)** $f(x)$ over $-\infty < x < \infty$, which has the following properties:

* (a) $f(x) \geq 0$ for $-\infty < x < \infty$

```python
import math

def law_of_total_probability(partitions, conditional_probabilities):
    """
    Calculates the total probability of an event using the law of total probability.

    Args:
        partitions: A list of probabilities of the partition events (A_i).
        conditional_probabilities: A list of conditional probabilities P(B|A_i).

    Returns:
        The total probability of event B.
    """

    if len(partitions) != len(conditional_probabilities):
        raise ValueError("Number of partitions and conditional probabilities must be the same.")

    total_probability = 0
    for i in range(len(partitions)):
        total_probability += partitions[i] * conditional_probabilities[i]

    return total_probability

def discrete_random_variable_pmf(values, probabilities):
    """
    Creates a probability mass function (PMF) for a discrete random variable.

    Args:
        values: A list of possible values of the random variable (x_i).
        probabilities: A list of corresponding probabilities (p(x_i)).

    Returns:
        A dictionary representing the PMF, where keys are values and values are probabilities.
    """

    if len(values) != len(probabilities):
        raise ValueError("Number of values and probabilities must be the same.")

    pmf = dict(zip(values, probabilities))
    return pmf

def calculate_probability_of_event(pmf, event_condition):
    """
    Calculates the probability of an event for a discrete random variable given its PMF.

    Args:
        pmf: A dictionary representing the PMF.
        event_condition: A function that defines the event condition (e.g., lambda x: x <= 2).

    Returns:
        The probability of the event.
    """

    probability = 0
    for value, prob in pmf.items():
        if event_condition(value):
            probability += prob
    return probability

def geometric_pmf(n, p):
    """
    Calculates the probability mass function for a geometric distribution.

    Args:
        n: The number of trials until the first success.
        p: The probability of success on each trial.

    Returns:
        The probability P(N = n).
    """
    if n < 1 or p <= 0 or p > 1:
        raise ValueError("Invalid input values.")
    return (1 - p)**(n - 1) * p

def geometric_cdf(k, p):
    """
    Calculates the cumulative distribution function for a geometric distribution.

    Args:
        k: The maximum number of trials.
        p: The probability of success on each trial.

    Returns:
        The probability P(N <= k).
    """
    if k < 1 or p <= 0 or p > 1:
        raise ValueError("Invalid input values.")
    return 1 - (1 - p)**k

# Example usage for geometric distribution (Example 1.5):
p_success = 1/6  # Probability of rolling a 6

n_trials = 3
prob_n_trials = geometric_pmf(n_trials, p_success)
print(f"Probability of first 6 on {n_trials}th throw: {prob_n_trials}")

k_trials = 4
prob_k_trials = geometric_cdf(k_trials, p_success)
print(f"Probability of first 6 within {k_trials} throws: {prob_k_trials}")

# Example usage for law of total probability:
partitions = [0.3, 0.4, 0.3]
conditional_probabilities = [0.1, 0.5, 0.8]
total_prob = law_of_total_probability(partitions, conditional_probabilities)
print(f"Total probability: {total_prob}")

# Example usage for discrete random variable PMF and calculating event probability:
values = [0, 1, 2, 3]
probabilities = [0.1, 0.2, 0.3, 0.4]
pmf = discrete_random_variable_pmf(values, probabilities)
print(f"PMF: {pmf}")

event_probability = calculate_probability_of_event(pmf, lambda x: x <= 2)
print(f"Probability of X <= 2: {event_probability}")

event_probability_equal_1 = calculate_probability_of_event(pmf, lambda x: x == 1)
print(f"Probability of X == 1: {event_probability_equal_1}")
```

## Example 1.6: Uniform Distribution

Show that
$$f(x) = \begin{cases} \frac{1}{b-a} & a \leq x \leq b \\ 0 & \text{otherwise} \end{cases}$$
is a possible probability density function (PDF). Find its cumulative distribution function (CDF).

**Solution:**

To show that $f(x)$ is a PDF, it must satisfy the conditions:

* (a) $f(x) \geq 0$ for all $x$
* (c) $\int_{-\infty}^{\infty} f(x) dx = 1$

Clearly, $f(x) \geq 0$ since $\frac{1}{b-a} \geq 0$ for $a \leq x \leq b$ and $f(x) = 0$ otherwise.

Now, let's verify condition (c):

$$\int_{-\infty}^{\infty} f(x) dx = \int_{a}^{b} \frac{1}{b-a} dx = \frac{1}{b-a} \int_{a}^{b} dx = \frac{1}{b-a} [x]_{a}^{b} = \frac{1}{b-a} (b-a) = 1$$

Thus, $f(x)$ is a valid PDF.

Now, let's find the CDF $F(x)$:

$$F(x) = \int_{-\infty}^{x} f(u) du$$

* For $x < a$:
    $$F(x) = \int_{-\infty}^{x} 0 du = 0$$
* For $a \leq x \leq b$:
    $$F(x) = \int_{-\infty}^{a} 0 du + \int_{a}^{x} \frac{1}{b-a} du = \frac{1}{b-a} [u]_{a}^{x} = \frac{x-a}{b-a}$$
* For $x > b$:
    $$F(x) = \int_{-\infty}^{a} 0 du + \int_{a}^{b} \frac{1}{b-a} du + \int_{b}^{x} 0 du = 1$$

Therefore, the CDF is:

$$F(x) = \begin{cases} 0 & x < a \\ \frac{x-a}{b-a} & a \leq x \leq b \\ 1 & x > b \end{cases}$$

This $f(x)$ is the PDF of the uniform distribution.

## 1.6 Mean and Variance

The **mean** (or **expectation** or **expected value**), $E(X)$ or $\mu$, of a discrete random variable $X$ is:

$$\mu = E(X) = \sum_{i=0}^{\infty} x_i p(x_i)$$

where $p(x_i) = P(X = x_i)$.

For a continuous random variable $X$, the mean is:

$$\mu = E(X) = \int_{-\infty}^{\infty} x f(x) dx$$

where $f(x)$ is the PDF.

The mean represents the weighted average of the values of $X$.

The **variance** of $X$, denoted by $V(X)$ or $\sigma^2$, measures the spread or dispersion of the probability distribution of $X$:

$$\sigma^2 = V(X) = E[(X - E(X))^2] = E[(X - \mu)^2]$$

* For a discrete random variable $X$:
    $$\sigma^2 = \sum_{i=0}^{\infty} (x_i - \mu)^2 p(x_i) \quad \text{or} \quad \sigma^2 = \sum_{x \in V} (x - \mu)^2 p(x)$$
* For a continuous random variable $X$:
    $$\sigma^2 = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx$$

The variance is the mean of the squared deviations of each value of $X$ from the mean.





```python
import math

def uniform_pdf(x, a, b):
    """
    Calculates the probability density function (PDF) for a uniform distribution.

    Args:
        x: The value at which to evaluate the PDF.
        a: The lower bound of the distribution.
        b: The upper bound of the distribution.

    Returns:
        The PDF value at x.
    """
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

def uniform_cdf(x, a, b):
    """
    Calculates the cumulative distribution function (CDF) for a uniform distribution.

    Args:
        x: The value at which to evaluate the CDF.
        a: The lower bound of the distribution.
        b: The upper bound of the distribution.

    Returns:
        The CDF value at x.
    """
    if x < a:
        return 0
    elif a <= x <= b:
        return (x - a) / (b - a)
    else:
        return 1

def discrete_mean(values, probabilities):
    """
    Calculates the mean (expected value) of a discrete random variable.

    Args:
        values: A list of possible values of the random variable.
        probabilities: A list of corresponding probabilities.

    Returns:
        The mean of the random variable.
    """
    if len(values) != len(probabilities):
        raise ValueError("Number of values and probabilities must be the same.")
    return sum(v * p for v, p in zip(values, probabilities))

def continuous_mean(pdf, lower_bound, upper_bound, num_samples=1000):
    """
    Approximates the mean of a continuous random variable using numerical integration.

    Args:
        pdf: The probability density function.
        lower_bound: The lower bound of the integration interval.
        upper_bound: The upper bound of the integration interval.
        num_samples: The number of samples to use for numerical integration.

    Returns:
        The approximate mean of the random variable.
    """
    delta = (upper_bound - lower_bound) / num_samples
    mean = 0
    for i in range(num_samples):
        x = lower_bound + i * delta + delta / 2  # Midpoint rule
        mean += x * pdf(x) * delta
    return mean

def discrete_variance(values, probabilities):
    """
    Calculates the variance of a discrete random variable.

    Args:
        values: A list of possible values of the random variable.
        probabilities: A list of corresponding probabilities.

    Returns:
        The variance of the random variable.
    """
    mean = discrete_mean(values, probabilities)
    return sum((v - mean) ** 2 * p for v, p in zip(values, probabilities))

def continuous_variance(pdf, lower_bound, upper_bound, num_samples=1000):
    """
    Approximates the variance of a continuous random variable using numerical integration.

    Args:
        pdf: The probability density function.
        lower_bound: The lower bound of the integration interval.
        upper_bound: The upper bound of the integration interval.
        num_samples: The number of samples to use for numerical integration.

    Returns:
        The approximate variance of the random variable.
    """
    mean = continuous_mean(pdf, lower_bound, upper_bound, num_samples)
    delta = (upper_bound - lower_bound) / num_samples
    variance = 0
    for i in range(num_samples):
        x = lower_bound + i * delta + delta / 2
        variance += (x - mean) ** 2 * pdf(x) * delta
    return variance

# Example usage for uniform distribution:
a = 1
b = 5
x_val = 3
print(f"Uniform PDF at x={x_val}: {uniform_pdf(x_val, a, b)}")
print(f"Uniform CDF at x={x_val}: {uniform_cdf(x_val, a, b)}")
print(f"Uniform mean: {continuous_mean(lambda x: uniform_pdf(x,a,b), a, b)}")
print(f"Uniform variance: {continuous_variance(lambda x: uniform_pdf(x,a,b), a, b)}")

# Example usage for discrete mean and variance:
values = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]
print(f"Discrete mean: {discrete_mean(values, probabilities)}")
print(f"Discrete variance: {discrete_variance(values, probabilities)}")
```

##  Mean and Variance (Continued)

The **standard deviation (sd)** of $X$, denoted by $\sigma$, is the square root of the variance:

$$\sigma = \sqrt{V(X)}$$

This provides a measure of variation in the same units as the mean.

A function of a random variable, $h(X)$, is itself a random variable. The expectation of $h(X)$ is:

* For a discrete random variable $X$:
    $$E[h(X)] = \sum_{i=0}^{\infty} h(x_i) p(x_i) \quad \text{or} \quad E[h(X)] = \sum_{x \in V} h(x) p(x)$$
* For a continuous random variable $X$:
    $$E[h(X)] = \int_{-\infty}^{\infty} h(x) f(x) dx$$

For a linear function of $X$, $aX + b$, where $a$ and $b$ are constants:

* Expectation:
    $$E(aX + b) = aE(X) + b = a\mu + b$$
* Variance:
    $$V(aX + b) = E[(aX + b - a\mu - b)^2] = E[(aX - a\mu)^2] = a^2 E[(X - \mu)^2] = a^2 V(X)$$

Note that adding a constant $b$ does not affect the variance.

The variance can also be calculated using the computational formula:

$$V(X) = E[(X - \mu)^2] = E(X^2) - 2\mu E(X) + \mu^2 = E(X^2) - \mu^2$$

These results hold for both discrete and continuous random variables.

For expectations of a linear combination of functions of $X$:

$$E\left[\sum_{i=1}^{k} a_i h_i(X)\right] = \sum_{i=1}^{k} a_i E[h_i(X)]$$

where $a_i$ are constants and $h_i(X)$ are functions of $X$.

##  Some Standard Discrete Probability Distributions

In this section, we will look at discrete random variables $X$ with probability mass function $p(x_i)$ or $p_i$, where $x_i$ takes integer values.

Each of the random variables considered are numerical outcomes of independent repetitions (or trials) of a simple experiment known as a **Bernoulli experiment**. This is an experiment where there are only two possible outcomes, often referred to as "success" and "failure".

```python
import math

def discrete_mean(values, probabilities):
    """Calculates the mean (expected value) of a discrete random variable."""
    if len(values) != len(probabilities):
        raise ValueError("Number of values and probabilities must be the same.")
    return sum(v * p for v, p in zip(values, probabilities))

def discrete_variance(values, probabilities):
    """Calculates the variance of a discrete random variable."""
    mean = discrete_mean(values, probabilities)
    return sum((v - mean) ** 2 * p for v, p in zip(values, probabilities))

def discrete_standard_deviation(values, probabilities):
    """Calculates the standard deviation of a discrete random variable."""
    return math.sqrt(discrete_variance(values, probabilities))

def expected_value_h_x_discrete(values, probabilities, h_function):
    """Calculates E[h(X)] for a discrete random variable."""
    if len(values) != len(probabilities):
        raise ValueError("Number of values and probabilities must be the same.")
    return sum(h_function(v) * p for v, p in zip(values, probabilities))

def expected_a_x_plus_b_discrete(a, b, values, probabilities):
    """Calculates E[aX + b] for a discrete random variable."""
    mean = discrete_mean(values, probabilities)
    return a * mean + b

def variance_a_x_plus_b_discrete(a, values, probabilities):
    """Calculates V[aX + b] for a discrete random variable."""
    variance = discrete_variance(values, probabilities)
    return a**2 * variance

def variance_e_x_squared_minus_mu_squared_discrete(values, probabilities):
    """Calculates V(X) using E(X^2) - mu^2 for a discrete random variable."""
    mean = discrete_mean(values, probabilities)
    e_x_squared = expected_value_h_x_discrete(values, probabilities, lambda x: x**2)
    return e_x_squared - mean**2

def expected_linear_combination_h_x_discrete(a_list, h_functions, values, probabilities):
    """Calculates E[sum(a_i * h_i(X))] for a discrete random variable."""
    if len(a_list) != len(h_functions):
        raise ValueError("Number of a_i and h_i(X) must be the same.")
    expected_sum = 0
    for a, h in zip(a_list, h_functions):
        expected_sum += a * expected_value_h_x_discrete(values, probabilities, h)
    return expected_sum

# Example Usage:

values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

print(f"Mean: {discrete_mean(values, probabilities)}")
print(f"Variance: {discrete_variance(values, probabilities)}")
print(f"Standard Deviation: {discrete_standard_deviation(values, probabilities)}")

h_function = lambda x: x**3
print(f"E[h(X)]: {expected_value_h_x_discrete(values, probabilities, h_function)}")

a = 2
b = 3
print(f"E[aX + b]: {expected_a_x_plus_b_discrete(a, b, values, probabilities)}")
print(f"V[aX + b]: {variance_a_x_plus_b_discrete(a, values, probabilities)}")
print(f"V(X) (E[X^2] - mu^2): {variance_e_x_squared_minus_mu_squared_discrete(values, probabilities)}")

a_list = [1, 2, 3]
h_functions = [lambda x: x, lambda x: x**2, lambda x: x**3]
print(f"E[sum(a_i * h_i(X))]: {expected_linear_combination_h_x_discrete(a_list, h_functions, values, probabilities)}")
```

The probability function of this random variable is:

$$p_k = P(X = k) = \binom{n}{k} p^k q^{n-k}, \quad k = 0, 1, 2, ..., n$$

This is the **binomial distribution** with parameters $n$ and $p$.

The mean and variance of the binomial distribution are $np$ and $npq$, respectively.

This is the **geometric distribution** with parameter $p$.

This process is known as **inverse sampling**. Note that successive trials are independent.

```python
import math

def bernoulli_pmf(x, p):
    """Calculates the probability mass function (PMF) for a Bernoulli distribution."""
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0

def bernoulli_mean(p):
    """Calculates the mean of a Bernoulli distribution."""
    return p

def bernoulli_variance(p):
    """Calculates the variance of a Bernoulli distribution."""
    return p * (1 - p)

def binomial_pmf(k, n, p):
    """Calculates the PMF for a binomial distribution."""
    if k < 0 or k > n:
        return 0
    combination = math.comb(n, k)
    return combination * (p**k) * ((1 - p)**(n - k))

def binomial_mean(n, p):
    """Calculates the mean of a binomial distribution."""
    return n * p

def binomial_variance(n, p):
    """Calculates the variance of a binomial distribution."""
    return n * p * (1 - p)

def geometric_pmf(k, p):
    """Calculates the PMF for a geometric distribution."""
    if k < 1:
        return 0
    return ((1 - p)**(k - 1)) * p

def geometric_mean(p):
    """Calculates the mean of a geometric distribution."""
    return 1 / p

def geometric_variance(p):
    """Calculates the variance of a geometric distribution."""
    return (1 - p) / (p**2)

# Example Usage:

# Bernoulli Distribution:
p_bernoulli = 0.6
print(f"Bernoulli PMF (X=1): {bernoulli_pmf(1, p_bernoulli)}")
print(f"Bernoulli PMF (X=0): {bernoulli_pmf(0, p_bernoulli)}")
print(f"Bernoulli Mean: {bernoulli_mean(p_bernoulli)}")
print(f"Bernoulli Variance: {bernoulli_variance(p_bernoulli)}")

# Binomial Distribution:
n_binomial = 5
p_binomial = 0.4
k_binomial = 3
print(f"Binomial PMF (k={k_binomial}): {binomial_pmf(k_binomial, n_binomial, p_binomial)}")
print(f"Binomial Mean: {binomial_mean(n_binomial, p_binomial)}")
print(f"Binomial Variance: {binomial_variance(n_binomial, p_binomial)}")

# Geometric Distribution:
p_geometric = 0.2
k_geometric = 4
print(f"Geometric PMF (k={k_geometric}): {geometric_pmf(k_geometric, p_geometric)}")
print(f"Geometric Mean: {geometric_mean(p_geometric)}")
print(f"Geometric Variance: {geometric_variance(p_geometric)}")
```

## 1.7 Some Standard Discrete Probability Distributions (Continued)

The probabilities in the geometric distribution form a geometric series with common ratio $q = 1 - p$. The sample space is countably infinite.

The mean of the geometric distribution is $\mu = \frac{1}{p}$, and the variance is $\sigma^2 = \frac{q}{p^2}$.

The geometric distribution possesses the **"no memory" property**:

$$P(X > a + b | X > a) = P(X > b)$$

where $a$ and $b$ are positive integers.

This means that if an event has not occurred in the first $a$ trials, the probability that it will occur in the next $b$ trials is the same as in the first $b$ trials.

Proof:

$$P(X > a + b | X > a) = \frac{P(X > a + b \cap X > a)}{P(X > a)} = \frac{P(X > a + b)}{P(X > a)}$$

Since $P(X > x) = q^x$:

$$P(X > a + b | X > a) = \frac{q^{a+b}}{q^a} = q^b = P(X > b)$$

The converse is also true.

Now, consider the case where $r$ ($> 1$) successes have occurred. The probability function of the number of trials is:

$$p_k = \binom{k-1}{r-1} p^r q^{k-r}, \quad k = r, r+1, ...$$

This is the **Pascal** or **negative binomial distribution** with parameters $r$ and $p$.

The mean is $\frac{r}{p}$, and the variance is $\frac{rq}{p^2}$.

The Pascal distribution is related to the geometric distribution in the same way that the binomial distribution is related to the Bernoulli distribution.

The binomial random variable arises from observing $n$ independent Bernoulli random variables, and the Pascal arises from observing $r$ sets of geometric random variables.

Certain problems involve counting the number of events that have occurred in a fixed time period, such as the number of alpha particle emissions or the number of customers in a queue. The **Poisson distribution** is appropriate for modeling these counts when the underlying process is completely random.

```python
import math

def geometric_pmf(k, p):
    """Calculates the PMF for a geometric distribution."""
    if k < 1:
        return 0
    return ((1 - p)**(k - 1)) * p

def geometric_mean(p):
    """Calculates the mean of a geometric distribution."""
    return 1 / p

def geometric_variance(p):
    """Calculates the variance of a geometric distribution."""
    return (1 - p) / (p**2)

def pascal_pmf(k, r, p):
    """Calculates the PMF for a Pascal (negative binomial) distribution."""
    if k < r:
        return 0

    def combinations(n, k):
        """Helper function to calculate combinations (n choose k)."""
        if k < 0 or k > n:
            return 0
        if k == 0 or k == n:
            return 1
        if k > n // 2:
            k = n - k
        result = 1
        for i in range(k):
            result = result * (n - i) // (i + 1)
        return result

    combination = combinations(k - 1, r - 1)
    return combination * (p**r) * ((1 - p)**(k - r))

def pascal_mean(r, p):
    """Calculates the mean of a Pascal distribution."""
    return r / p

def pascal_variance(r, p):
    """Calculates the variance of a Pascal distribution."""
    return r * (1 - p) / (p**2)

def poisson_pmf(k, lambda_val):
    """Calculates the PMF for a Poisson distribution."""
    if k < 0:
        return 0
    return (lambda_val**k * math.exp(-lambda_val)) / math.factorial(k)

def poisson_mean(lambda_val):
    """Calculates the mean of a Poisson distribution."""
    return lambda_val

def poisson_variance(lambda_val):
    """Calculates the variance of a Poisson distribution."""
    return lambda_val

# Example Usage:

# Geometric Distribution:
p_geometric = 0.2
k_geometric = 4
print(f"Geometric PMF (k={k_geometric}): {geometric_pmf(k_geometric, p_geometric)}")
print(f"Geometric Mean: {geometric_mean(p_geometric)}")
print(f"Geometric Variance: {geometric_variance(p_geometric)}")

# Pascal (Negative Binomial) Distribution:
r_pascal = 3
p_pascal = 0.4
k_pascal = 7
print(f"Pascal PMF (k={k_pascal}): {pascal_pmf(k_pascal, r_pascal, p_pascal)}")
print(f"Pascal Mean: {pascal_mean(r_pascal, p_pascal)}")
print(f"Pascal Variance: {pascal_variance(r_pascal, p_pascal)}")

# Poisson Distribution:
lambda_poisson = 2.5
k_poisson = 3
print(f"Poisson PMF (k={k_poisson}): {poisson_pmf(k_poisson, lambda_poisson)}")
print(f"Poisson Mean: {poisson_mean(lambda_poisson)}")
print(f"Poisson Variance: {poisson_variance(lambda_poisson)}")
```

## 1.8 Some Standard Continuous Probability Distributions

The **exponential distribution** is a continuous distribution used to model the time until a particular event occurs. In a Poisson process, it models the time between successive events. It can model inter-arrival times in queues or the lifetime of a component with a constant failure rate.

The probability density function (PDF) is:

$$f(x) = \begin{cases} \alpha e^{-\alpha x}, & x \geq 0 \\ 0, & x < 0 \end{cases}$$

where $\alpha$ is a positive parameter.

It is a valid PDF because:

$$\int_{-\infty}^{\infty} f(x) dx = \int_{0}^{\infty} \alpha e^{-\alpha x} dx = 1$$

The mean and variance of the exponential distribution are:

$$\mu = E(X) = \int_{0}^{\infty} \alpha x e^{-\alpha x} dx = \frac{1}{\alpha}$$

```python
import math

def poisson_pmf(k, lambda_val):
    """Calculates the PMF for a Poisson distribution."""
    if k < 0:
        return 0
    return (lambda_val**k * math.exp(-lambda_val)) / math.factorial(k)

def poisson_mean(lambda_val):
    """Calculates the mean of a Poisson distribution."""
    return lambda_val

def poisson_variance(lambda_val):
    """Calculates the variance of a Poisson distribution."""
    return lambda_val

def discrete_uniform_pmf(k, r, n):
    """Calculates the PMF for a discrete uniform distribution."""
    if r <= k < r + n:
        return 1 / n
    else:
        return 0

def discrete_uniform_mean(r, n):
    """Calculates the mean of a discrete uniform distribution."""
    return r + (n - 1) / 2

def discrete_uniform_variance(r, n):
    """Calculates the variance of a discrete uniform distribution."""
    return (n**2 - 1) / 12

def exponential_pdf(x, alpha):
    """Calculates the PDF for an exponential distribution."""
    if x >= 0:
        return alpha * math.exp(-alpha * x)
    else:
        return 0

def exponential_mean(alpha):
    """Calculates the mean of an exponential distribution."""
    return 1 / alpha

def exponential_variance(alpha):
    """Calculates the variance of an exponential distribution."""
    return 1 / (alpha**2)

# Example Usage:

# Poisson Distribution:
lambda_poisson = 2.5
k_poisson = 3
print(f"Poisson PMF (k={k_poisson}): {poisson_pmf(k_poisson, lambda_poisson)}")
print(f"Poisson Mean: {poisson_mean(lambda_poisson)}")
print(f"Poisson Variance: {poisson_variance(lambda_poisson)}")

# Discrete Uniform Distribution:
r_uniform = 2
n_uniform = 5
k_uniform = 4
print(f"Discrete Uniform PMF (k={k_uniform}): {discrete_uniform_pmf(k_uniform, r_uniform, n_uniform)}")
print(f"Discrete Uniform Mean: {discrete_uniform_mean(r_uniform, n_uniform)}")
print(f"Discrete Uniform Variance: {discrete_uniform_variance(r_uniform, n_uniform)}")

# Exponential Distribution:
alpha_exponential = 0.5
x_exponential = 2
print(f"Exponential PDF (x={x_exponential}): {exponential_pdf(x_exponential, alpha_exponential)}")
print(f"Exponential Mean: {exponential_mean(alpha_exponential)}")
print(f"Exponential Variance: {exponential_variance(alpha_exponential)}")
```

## 1.8 Some Standard Continuous Probability Distributions (Continued)

The variance of the exponential distribution is:

$$\sigma^2 = V(X) = \int_{0}^{\infty} \left(x - \frac{1}{\alpha}\right)^2 \alpha e^{-\alpha x} dx = \frac{1}{\alpha^2}$$

The cumulative distribution function (CDF) $F(x)$ is:

$$F(x) = \int_{-\infty}^{x} f(u) du = \int_{0}^{x} \alpha e^{-\alpha y} dy = 1 - e^{-\alpha x}$$

For an exponentially distributed random variable $X$, the "no-memory" property holds:

$$P(X > a + b | X > a) = \frac{P(X > a + b)}{P(X > a)} = \frac{e^{-\alpha(a+b)}}{e^{-\alpha a}} = e^{-\alpha b}$$

where $a, b > 0$. This result is independent of $a$.

The exponential distribution is the only continuous distribution with this property.

A random variable $X$ has a **normal distribution** with mean $\mu$ and variance $\sigma^2$ if its PDF is:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad -\infty < x < \infty$$

This is denoted as $N(\mu, \sigma^2)$.

It can be shown that:

$$\int_{-\infty}^{\infty} f(x) dx = 1$$

And that:

$$E(X) = \mu, \quad V(X) = \sigma^2$$

The **standard normal distribution**, $N(0, 1)$, is obtained by standardizing $X$ using $Z = \frac{X - \mu}{\sigma}$:

$$f(z) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{z^2}{2}\right), \quad -\infty < z < \infty$$

This transformation allows us to calculate probabilities for any normally distributed random variable.

For example, if $X \sim N(2, 9)$, then:

$$P(-1 < X < 3) = P\left(\frac{-1-2}{3} < Z < \frac{3-2}{3}\right) = P(-1 < Z < \frac{1}{3}) = \Phi\left(\frac{1}{3}\right) - \Phi(-1) \approx 0.629 - 0.159 = 0.470$$

where $\Phi(z)$ is the cumulative distribution function of the standard normal distribution.

```python
import math

def exponential_pdf(x, alpha):
    """Calculates the PDF for an exponential distribution."""
    if x >= 0:
        return alpha * math.exp(-alpha * x)
    else:
        return 0

def exponential_mean(alpha):
    """Calculates the mean of an exponential distribution."""
    return 1 / alpha

def exponential_variance(alpha):
    """Calculates the variance of an exponential distribution."""
    return 1 / (alpha**2)

def exponential_cdf(x, alpha):
    """Calculates the CDF for an exponential distribution."""
    if x >= 0:
        return 1 - math.exp(-alpha * x)
    else:
        return 0

def standard_normal_pdf(z):
    """Calculates the PDF for a standard normal distribution."""
    return (1 / math.sqrt(2 * math.pi)) * math.exp(-0.5 * z**2)

def normal_pdf(x, mu, sigma):
    """Calculates the PDF for a normal distribution."""
    return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mu) / sigma)**2)

def standard_normal_cdf(z):
    """Approximates the CDF for a standard normal distribution."""
    # Approximation using error function (erf)
    return 0.5 * (1 + math.erf(z / math.sqrt(2)))

def normal_cdf(x, mu, sigma):
    """Calculates the CDF for a normal distribution."""
    z = (x - mu) / sigma
    return standard_normal_cdf(z)
# Example Usage:

# Exponential Distribution:
alpha_exponential = 0.5
x_exponential = 2
print(f"Exponential PDF (x={x_exponential}): {exponential_pdf(x_exponential, alpha_exponential)}")
print(f"Exponential Mean: {exponential_mean(alpha_exponential)}")
print(f"Exponential Variance: {exponential_variance(alpha_exponential)}")
print(f"Exponential CDF (x={x_exponential}): {exponential_cdf(x_exponential, alpha_exponential)}")

# Standard Normal Distribution:
z_standard = 1
print(f"Standard Normal PDF (z={z_standard}): {standard_normal_pdf(z_standard)}")
print(f"Standard Normal CDF (z={z_standard}): {standard_normal_cdf(z_standard)}")

# Normal Distribution:
mu_normal = 2
sigma_normal = 3
x_normal = 3
print(f"Normal PDF (x={x_normal}): {normal_pdf(x_normal, mu_normal, sigma_normal)}")
print(f"Normal CDF (x={x_normal}): {normal_cdf(x_normal, mu_normal, sigma_normal)}")

#Example to calculate P(-1 < X < 3) for X~N(2,9)
p_neg1_to_3 = normal_cdf(3, 2, 3) - normal_cdf(-1,2,3)
print(f"P(-1 < X < 3) where X~N(2,9) = {p_neg1_to_3}")
```
