
# Linear Neural Networks for Regression

## Introduction

Linear neural networks for regression are a class of shallow neural networks where the inputs connect directly to the outputs. These networks are important for several reasons. First, they allow us to focus on the basics of neural network training, including parameterizing the output layer, handling data, specifying a loss function, and training the model. Second, this class of shallow networks includes linear models, which encompass many classical methods for statistical prediction, such as linear and softmax regression. Understanding these classical tools is crucial as they are widely used and serve as baselines for justifying the use of more complex architectures.

## Regression Problems

Regression problems arise when we want to predict a numerical value. Examples include predicting prices (of homes, stocks, etc.), predicting the length of stay for patients in the hospital, forecasting demand for retail sales, and many others. Not all prediction problems are classic regression problems, as we will introduce classification problems later, where the goal is to predict membership among a set of categories.

## Running Example: House Price Prediction

As a running example, suppose we want to estimate the prices of houses (in dollars) based on their area (in square feet) and age (in years). To develop a model for predicting house prices, we need a dataset containing sales data, including the sales price, area, and age for each home. In machine learning terminology, this dataset is called a training dataset or training set, and each row (containing data for one sale) is called an example, data point, instance, or sample. The thing we are trying to predict (price) is called a label or target. The variables (age and area) upon which the predictions are based are called features or covariates.

## Linear Regression

Linear regression is one of the simplest and most popular tools for tackling regression problems. It dates back to the early 19th century (Gauss, 1809, Legendre, 1805) and is based on a few simple assumptions. First, we assume that the relationship between features $x$ and target $y$ is approximately linear, i.e., the conditional mean $E[Y | X = x]$ can be expressed as a weighted sum of the features $x$. This setup allows for the target value to deviate from its expected value due to observation noise. Next, we assume that the noise follows a Gaussian distribution.

## Model

At the heart of linear regression is a model that describes how features can be transformed into an estimate of the target. The assumption of linearity means that the expected value of the target (price) can be expressed as a weighted sum of the features (area and age):

$$
\text{price} = w_{\text{area}} \cdot \text{area} + w_{\text{age}} \cdot \text{age} + b
$$

Here, $w_{\text{area}}$ and $w_{\text{age}}$ are called weights, and $b$ is called a bias, offset, or intercept. The weights determine the influence of each feature on our prediction, and the bias determines the value of the estimate when all features are zero.

Given a dataset, the goal of linear regression is to find the weight vector $w$ and bias term $b$ that, on average, make our model's predictions fit the true prices observed in the data as closely as possible.

## Matrix-Vector Notation

When our inputs consist of $d$ features, we can assign each an index (between 1 and $d$) and express our prediction $\hat{y}$ as:

$$
\hat{y} = w_1 x_1 + ... + w_d x_d + b
$$

We can represent the features of our entire dataset of $n$ examples as a design matrix $X \in \mathbb{R}^{n \times d}$. Here, $X$ contains one row for every example and one column for every feature. For a collection of features $X$, the predictions $\hat{y} \in \mathbb{R}^n$ can be expressed via the matrix-vector product:

$$
\hat{y} = Xw + b
$$

## Quality Measure and Updating Procedure

Before we can search for the best parameters $w$ and $b$, we need two more things: (i) a quality measure for some given model, and (ii) a procedure for updating the model to improve its quality.

## Conclusion

Linear regression is a fundamental tool for tackling regression problems, and understanding it is crucial for developing more complex models. In the next chapter, we will extend our modeling repertoire by developing linear neural networks for classification.

````python
import torch

class LinearRegression:
    def __init__(self, input_size):
        self.input_size = input_size
        self.weights = torch.randn(input_size, requires_grad=True)
        self.bias = torch.randn(1, requires_grad=True)

    def forward(self, x):
        return torch.matmul(x, self.weights) + self.bias

# Example usage
input_size = 2  # Number of features
model = LinearRegression(input_size)

# Generate some synthetic data
X = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float32)
y = torch.tensor([3, 5, 7], dtype=torch.float32)

# Forward pass
y_pred = model.forward(X)

# Calculate the loss (mean squared error)
loss = torch.mean((y_pred - y) ** 2)

# Print the loss
print("Loss:", loss.item())
````



# Loss Function

## Introduction

A loss function is used to quantify the distance between the real and predicted values of the target in a regression problem. It provides a measure of how well the model is fitting the data. The loss is typically a non-negative number, where smaller values indicate better fit, and perfect predictions incur a loss of 0.

## Squared Error Loss

The most common loss function for regression problems is the squared error loss. For a predicted value $\hat{y}_i$ and the corresponding true label $y_i$, the squared error loss is given by:

$$
l(i) = \frac{1}{2}(\hat{y}_i - y_i)^2
$$

The constant $\frac{1}{2}$ is included for notational convenience, as it cancels out when taking the derivative of the loss.

## Empirical Error

The empirical error is the average loss over the entire dataset of $n$ examples:

$$
L(w, b) = \frac{1}{n} \sum_{i=1}^n \frac{1}{2}(\hat{y}_i - y_i)^2
$$

## Minimizing the Loss

When training the model, the goal is to find the parameters $(w^*, b^*)$ that minimize the total loss across all training examples:

$$
w^*, b^* = \arg\min L(w, b)
$$

## Analytic Solution for Linear Regression

Surprisingly, linear regression presents us with an easy optimization problem that admits an analytic solution. By appending a column of ones to the design matrix $X$ to subsume the bias $b$ into the parameter $w$, we can minimize the squared error loss:

$$
\min_w \|y - Xw\|^2
$$

If the design matrix $X$ has full rank (no feature is linearly dependent on the others), there will be a unique critical point on the loss surface that corresponds to the minimum of the loss over the entire domain. Taking the derivative of the loss with respect to $w$ and setting it equal to zero yields:

$$
X^\top y = X^\top Xw
$$

Solving for $w$ gives us the optimal solution:

$$
w^* = (X^\top X)^{-1} X^\top y
$$

This solution is unique when the matrix $X^\top X$ is invertible, i.e., when the columns of the design matrix are linearly independent.

## Conclusion

While linear regression provides an analytic solution for parameter optimization, this is not the case for most models we will encounter in deep learning. Analytic solutions are rare and often exclude exciting aspects of deep learning. However, understanding the analytic solution for linear regression is valuable for its simplicity and mathematical analysis.

````python
import torch

class LinearRegression:
    def __init__(self, input_size):
        self.input_size = input_size
        self.weights = torch.randn(input_size, requires_grad=True)
        self.bias = torch.randn(1, requires_grad=True)

    def forward(self, x):
        return torch.matmul(x, self.weights) + self.bias

# Example usage
input_size = 2  # Number of features
model = LinearRegression(input_size)

# Generate some synthetic data
X = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float32)
y = torch.tensor([3, 5, 7], dtype=torch.float32)

# Forward pass
y_pred = model.forward(X)

# Calculate the loss (mean squared error)
loss = torch.mean((y_pred - y) ** 2)

# Print the loss
print("Loss:", loss.item())

# Analytic solution for parameter optimization
X_transpose = X.t()
X_transpose_X = torch.matmul(X_transpose, X)
X_transpose_y = torch.matmul(X_transpose, y)
weights = torch.matmul(torch.inverse(X_transpose_X), X_transpose_y)
bias = torch.mean(y - torch.matmul(X, weights))

# Print the optimized weights and bias
print("Optimized Weights:", weights.detach().numpy())
print("Optimized Bias:", bias.item())
````


# Minibatch Stochastic Gradient Descent

## Introduction

Minibatch stochastic gradient descent (SGD) is a widely used optimization algorithm in deep learning. It is an iterative algorithm that updates the model parameters by taking small steps in the direction that incrementally lowers the loss function. This algorithm is particularly useful when dealing with large datasets, as it allows for efficient updates by considering a subset of the data at each iteration.

## Algorithm

Minibatch SGD proceeds as follows:

1. Initialize the values of the model parameters, typically at random.
2. Iteratively sample random minibatches from the data.
3. Compute the derivative (gradient) of the average loss on the minibatch with respect to the model parameters.
4. Update the parameters by subtracting the gradient multiplied by a predetermined small positive value called the learning rate.

The update equation for minibatch SGD is:

$$
\begin{align*}
w &\leftarrow w - \eta \frac{1}{|B|} \sum_{i \in B} \frac{\partial l(i)}{\partial w} \\
b &\leftarrow b - \eta \frac{1}{|B|} \sum_{i \in B} \frac{\partial l(i)}{\partial b}
\end{align*}
$$

where $B$ is the minibatch, $|B|$ is the size of the minibatch, $i$ is the index of the example in the minibatch, $t$ is the iteration number, and $\eta$ is the learning rate.

## Advantages and Disadvantages

Minibatch SGD offers several advantages:

- It can be more computationally efficient than full-batch gradient descent, as it processes smaller subsets of data at a time.
- It can handle large datasets more effectively, as it does not require loading the entire dataset into memory.
- It can provide better generalization performance, as it introduces randomness in the parameter updates.

However, minibatch SGD also has some disadvantages:

- It may require careful tuning of hyperparameters, such as the minibatch size and learning rate.
- It can be sensitive to the choice of hyperparameters, and finding the optimal values may require experimentation.
- It may not converge to the global minimum of the loss function, as it is a stochastic optimization algorithm.

## Hyperparameters

Minibatch size and learning rate are hyperparameters that are typically user-defined. These hyperparameters can have a significant impact on the training process and the final model performance. They are often tuned manually or automatically using techniques like Bayesian optimization.

## Model Parameters

After training for a predetermined number of iterations or until a stopping criterion is met, the estimated model parameters, denoted $\hat{w}$ and $\hat{b}$, are recorded. These parameters represent the learned weights and bias of the linear regression model.

## Predictions

Given the model $\hat{w}^\top x + \hat{b}$, we can make predictions for new examples. For example, we can predict the sales price of a previously unseen house given its area $x_1$ and age $x_2$. In deep learning, the term "inference" is often used to refer to the prediction phase, but it is important to note that inference in statistics has a broader meaning, encompassing both parameter inference and label prediction.

## Conclusion

Minibatch SGD is a powerful optimization algorithm for training deep learning models. It provides a balance between computational efficiency and generalization performance. While it may require careful tuning of hyperparameters, it is widely used and has shown success in various deep learning applications.

````python
import torch

class LinearRegression:
    def __init__(self, input_size):
        self.input_size = input_size
        self.weights = torch.randn(input_size, requires_grad=True)
        self.bias = torch.randn(1, requires_grad=True)

    def forward(self, x):
        return torch.matmul(x, self.weights) + self.bias

# Example usage
input_size = 2  # Number of features
model = LinearRegression(input_size)

# Generate some synthetic data
X = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float32)
y = torch.tensor([3, 5, 7], dtype=torch.float32)

# Minibatch size
batch_size = 2

# Learning rate
learning_rate = 0.01

# Number of iterations
num_iterations = 1000

# Minibatch SGD
for _ in range(num_iterations):
    # Randomly sample a minibatch
    indices = torch.randperm(X.size(0))[:batch_size]
    minibatch_X = X[indices]
    minibatch_y = y[indices]

    # Forward pass
    y_pred = model.forward(minibatch_X)

    # Calculate the loss (mean squared error)
    loss = torch.mean((y_pred - minibatch_y) ** 2)

    # Backward pass
    loss.backward()

    # Update the parameters
    with torch.no_grad():
        self.weights -= learning_rate * self.weights.grad
        self.bias -= learning_rate * self.bias.grad

    # Zero the gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()

# Print the optimized weights and bias
print("Optimized Weights:", model.weights.detach().numpy())
print("Optimized Bias:", model.bias.item())
````

## Vectorization for Speed and The Normal Distribution in Regression

##  Vectorization for Speed

When training our models, we typically want to process whole minibatches of examples simultaneously. Doing this efficiently requires that we vectorize the calculations and leverage fast linear algebra libraries rather than writing costly for-loops in Python.

To illustrate why this matters so much, we can consider two methods for adding vectors. To start, we instantiate two 10,000-dimensional vectors containing all ones.

```python
import torch
import time

n = 10000
a = torch.ones(n)
b = torch.ones(n)
```

Now we can benchmark the workloads. First, we add them, one coordinate at a time, using a for-loop.

```python
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
f'{time.time() - t:.5f} sec'
```

```
'0.18086 sec'
```

Alternatively, we rely on the reloaded `+` operator to compute the elementwise sum.

```python
t = time.time()
d = a + b
f'{time.time() - t:.5f} sec'
```

```
'0.00015 sec'
```

The second method is dramatically faster than the first. Vectorizing code often yields order-of-magnitude speedups. Moreover, we push more of the mathematics to the library without the need to write as many calculations ourselves, reducing the potential for errors and increasing the portability of the code.

##  The Normal Distribution and Squared Loss

We can provide a more formal motivation for the squared loss objective by making probabilistic assumptions about the distribution of noise. The normal distribution (also called the Gaussian) is defined as:

$$
p(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp \left( -\frac{1}{2} \frac{(x - \mu)^2}{\sigma^2} \right)
$$

Below we define a function to compute the normal distribution:

```python
import math
import numpy as np

def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 * (x - mu)**2 / sigma**2)
```

We can now visualize the normal distributions:

```python
import matplotlib.pyplot as plt

x = np.arange(-7, 7, 0.01)
params = [(0, 1), (0, 2), (3, 1)]

plt.figure(figsize=(4.5, 2.5))
for mu, sigma in params:
    plt.plot(x, normal(x, mu, sigma), label=f'mean {mu}, std {sigma}')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

One way to motivate linear regression with squared loss is to assume that observations arise from noisy measurements, where the noise is normally distributed as follows:

$$
y = w^T x + b + \epsilon, \quad \epsilon \sim \mathcal{N}(0, \sigma^2)
$$

Thus, we can now write out the likelihood of seeing a particular $ y $ for a given $ x $ via:

$$
P(y \mid x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp \left( -\frac{1}{2} \frac{(y - w^T x - b)^2}{\sigma^2} \right)
$$

According to the principle of maximum likelihood, the best values of parameters $ w \$ and $ b $ are those that maximize the likelihood of the entire dataset:

$$
P(y \mid X) = \prod_{i=1}^{n} p(y^{(i)} \mid x^{(i)})
$$

By taking the negative log-likelihood, we obtain the loss function:

$$
- \log P(y \mid X) = \sum_{i=1}^{n} \left( \frac{1}{2} \log(2\pi\sigma^2) + \frac{(y^{(i)} - w^T x^{(i)} - b)^2}{2\sigma^2} \right)
$$

If we assume that $ \sigma $ is fixed, the first term can be ignored, leaving us with the squared error loss. This demonstrates that minimizing the mean squared error is equivalent to maximum likelihood estimation of a linear model under the assumption of additive Gaussian noise.




````python
import time
import torch
import math
import numpy as np
import matplotlib.pyplot as plt

# Vectorization for Speed
n = 10000
a = torch.ones(n)
b = torch.ones(n)

# Method 1: Using a for-loop
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
print(f'For-loop time: {time.time() - t:.5f} sec')

# Method 2: Using vectorized addition
t = time.time()
d = a + b
print(f'Vectorized time: {time.time() - t:.5f} sec')

# Normal Distribution
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 * (x - mu)**2 / sigma**2)

x = np.arange(-7, 7, 0.01)
params = [(0, 1), (0, 2), (3, 1)]

plt.figure(figsize=(6, 4))
for mu, sigma in params:
    plt.plot(x, normal(x, mu, sigma), label=f'mean {mu}, std {sigma}')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()

# Squared Loss
np.random.seed(0)
x_train = 2 * np.random.rand(100, 1)
y_train = 4 + 3 * x_train + np.random.randn(100, 1)

# Linear Regression using Normal Equation
X_b = np.c_[np.ones((100, 1)), x_train]  # add bias term
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_train)

print(f"Optimal parameters (w, b): {theta_best.ravel()}")

# Predictions
def predict(x, theta):
    return np.c_[np.ones((len(x), 1)), x].dot(theta)

x_test = np.array([[0], [2]])
y_pred = predict(x_test, theta_best)

plt.scatter(x_train, y_train, color='blue', label='Data')
plt.plot(x_test, y_pred, 'r-', label='Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

## Linear Regression as a Neural Network

While linear models are not sufficiently rich to express the many complicated neural networks that we will introduce in this book, neural networks are rich enough to subsume linear models as neural networks in which every feature is represented by an input neuron, all of which are connected directly to the output.

### Representation as a Neural Network

Linear regression can be visualized as a neural network where:
- The inputs are $ x_1, x_2, \dots, x_d $
- The output is $ o_1 $
- The model follows the equation:
  $$
  o_1 = \sum_{i=1}^{d} w_i x_i + b
  $$

Since we are predicting a single numerical value, there is only one output neuron. Linear regression is thus a single-layer fully connected neural network.

### Biological Analogy

The concept of artificial neural networks is inspired by biological neurons:
- **Dendrites** receive signals (input features $ x_i $).
- **Synapses** assign weights $ w_i $ to each input.
- **Cell body (Nucleus)** aggregates inputs as a weighted sum.
- **Axon terminals** pass the output to the next neuron.

In mathematical form, a biological neuron's processing can be approximated as:
  $$
  y = \sum_{i=1}^{d} w_i x_i + b
  $$
  followed by a nonlinear activation $ \sigma(y) $ (if applicable).

While neural networks were initially inspired by biological systems, modern deep learning research integrates concepts from multiple disciplines including mathematics, statistics, and computer science.

---

##  Summary

- **Linear regression** selects parameters to minimize squared loss on the training set.
- **Maximum likelihood estimation** motivates squared loss under Gaussian noise assumption.
- **Computational efficiency** considerations highlight the importance of vectorization.
- **Neural network representation** shows that linear regression is a one-layer neural network.
- **Generalization** to deep learning introduces core components like differentiability, optimization, and evaluation on unseen data.

Linear regression serves as a foundational concept before moving to more complex deep learning models.



````python
import torch
import torch.nn as nn
import torch.optim as optim

# Generate synthetic data
true_w = torch.tensor([2.0, -3.0])
true_b = 5.0
num_samples = 100

def synthetic_data(w, b, num_samples):
    X = torch.randn(num_samples, len(w))
    y = X @ w + b + 0.1 * torch.randn(num_samples)
    return X, y

X, y = synthetic_data(true_w, true_b, num_samples)

torch.manual_seed(42)

# Define the model as a single-layer neural network
class LinearRegression(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.linear = nn.Linear(input_dim, 1)
    
    def forward(self, x):
        return self.linear(x).squeeze(1)

# Instantiate model
model = LinearRegression(input_dim=2)

# Define loss function and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.03)

# Training loop
epochs = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    predictions = model(X)
    loss = loss_fn(predictions, y)
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')

# Print learned parameters
learned_w = model.linear.weight.data.numpy()
learned_b = model.linear.bias.data.numpy()
print(f'Learned weights: {learned_w}, Learned bias: {learned_b}')
````
