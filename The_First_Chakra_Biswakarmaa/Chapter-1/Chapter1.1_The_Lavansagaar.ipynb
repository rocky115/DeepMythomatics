## Here begins


The artificial neural networks discussed in this text are only remotely re- lated to their biological counterparts. In this section we will briefly describe those characteristics of brain function that have inspired the development of artificial neural networks. The brain consists of a large number (approximately 1011) of highly con- nected elements (approximately 104 connections per element) called neu- rons. For our purposes these neurons have three principal components: the dendrites, the cell body and the axon. The dendrites are tree-like receptive networks of nerve fibers that carry electrical signals into the cell body. The cell body effectively sums and thresholds these incoming signals. The axon is a single long fiber that carries the signal from the cell body out to other neurons. The point of contact between an axon of one cell and a dendrite of another cell is called a synapse. It is the arrangement of neurons and the strengths of the individual synapses, determined by a complex chemical process, that establishes the function of the neural network. Figure 1  Biological Neurons

Single-Input Neuron A single-input neuron is shown in Figure 2. The scalar input p is multi- plied by the scalar weight w to form wp , one of the terms that is sent to the summer. The other input, 1 , is multiplied by a bias b and then passed to the summer. The summer output n , often referred to as the net input, goes into a transfer function f , which produces the scalar neuron output a . (Some authors use the term activation function rather than transfer func- tion and offset rather than bias.) If we relate this simple model back to the biological neuron that we dis- cussed in Chapter 1, the weight w corresponds to the strength of a synapse, the cell body is represented by the summation and the transfer function, and the neuron output a represents the signal on the axon.

## Neuron Output Calculation

The neuron output can be calculated as:

$$ a = f(wp + b) $$

Where:
- $ w $ is the weight,
- $ p $ is the input, and
- $ b $ is the bias.

For example, if:

$$ w = 3, \, p = 2, \, b = -1.5 $$

Then the output is:

$$ a = f(3(2) - 1.5) = f(4.5) $$

The actual output $ a $ depends on the particular **transfer function** $ f $ that is chosen, which we will discuss in the next section.

### The Bias Term

The bias $ b $ behaves similarly to a weight, except it has a constant input of 1. If you do not wish to include a bias for a particular neuron, it can be omitted.

Note that both $ w $ and $ b $ are **adjustable scalar parameters** of the neuron. Typically, the **transfer function** is chosen by the designer, and the parameters $ w $ and $ b $ are adjusted using a **learning rule** so that the neuron’s input/output relationship meets a specific goal.

As described in the following section, there are different transfer functions suited for different purposes.

## Transfer Functions

The transfer function shown in **Figure 2.1** can be either a **linear** or a **nonlinear** function of \( n \). A specific transfer function is chosen to satisfy the requirements of the problem that the neuron is attempting to solve. Various transfer functions are discussed in this book, and three of the most commonly used ones are detailed below.

### Hard Limit Transfer Function

The **hard limit transfer function**, shown on the left side of **Figure 2**, sets the output of the neuron to 0 if the function argument is less than 0, and to 1 if the argument is greater than or equal to 0. Mathematically, it can be expressed as:

$$
f(n) =
\begin{cases}
0, & \text{if } n < 0 \\
1, & \text{if } n \geq 0
\end{cases}
$$

This function is commonly used to create neurons that classify inputs into two distinct categories. 



Figure 2. Hard Limit Transfer Function

## Hard Limit Transfer Function - Input/Output Characteristics

The graph on the right side of **Figure 2.2** illustrates the **input/output characteristic** of a single-input neuron using the hard limit transfer function. Here, the effect of the **weight** $ w $ and the **bias** $ b $ on the neuron’s output can be observed.

Additionally, an **icon** for the hard limit transfer function is placed between the two figures. These icons will replace the general $ f $ in network diagrams to indicate the specific transfer function being used.

### Linear Transfer Function

In contrast, the **linear transfer function** produces an output that is directly proportional to its input. Mathematically, this is expressed as:

$$
a = n
$$

This is shown in **Figure 3**.

## Log-Sigmoid Transfer Function

This transfer function takes the input, which may have any value between plus and minus infinity, and **squashes the output** into the range [0, 1]. The output is calculated using the following expression:

$$
a = \frac{1}{1 + e^{-n}} \tag{2.2}
$$

The **log-sigmoid transfer function** is commonly used in **multilayer networks** trained using the **backpropagation algorithm**, in part because this function is **differentiable** (see **Chapter 11**).

## Multiple-Input Neuron

Typically, a neuron has more than one input. A neuron with $ R $ inputs is illustrated in **Figure 5**. The individual inputs $ p_1, p_2, \dots, p_R $ are each weighted by the corresponding elements $ w_{1,1}, w_{1,2}, \dots, w_{1,R} $ of the weight matrix $ W $.

### Neuron Model


Figure 5. Multiple-Input Neuron

The neuron has a bias $ b $, which is summed with the weighted inputs to form the net input $ n $. This can be expressed as:

$$
n = w_{1,1} p_1 + w_{1,2} p_2 + \dots + w_{1,R} p_R + b \tag{2.3}
$$

This expression can also be written in **matrix form**:

$$
n = Wp + b \tag{2.4}
$$

where $ W $ is the weight matrix and $ p $ is the input vector. In the case of a single neuron, $ W $ will have only one row. Now the neuron output can be written as:

$$
a = f(Wp + b) \tag{2.5}
$$

### Indexing Convention

We have adopted a particular convention for assigning indices to the elements of the weight matrix. The **first index** indicates the destination neuron for that weight, while the **second index** indicates the source of the signal fed to the neuron. For example, in $ w_{1,2} $, this weight represents the connection to the **first neuron** from the **second input source**.

### Matrix Operations

Neural networks are often described using **matrices**, and this kind of matrix expression will be used throughout the book. If you're unfamiliar with matrix and vector operations, don't worry! We will review these topics in **Chapters 5 and 6**, and provide many examples and solved problems to explain the procedures in detail.


## Multiple-Input Neuron - Abbreviated Notation

When designing neural networks with several neurons, each with multiple inputs, and potentially more than one layer, the network can become complex. Drawing all the lines between inputs and neurons could obscure the main features. Therefore, we use an **abbreviated notation** to simplify the diagrams.

### Abbreviated Notation for a Neuron with $ R $ Inputs

   
Figure 6. Neuron with R Inputs, Abbreviated Notation

In **Figure 6**, a multiple-input neuron is represented using abbreviated notation:

- The input vector $ p $ is represented by a solid vertical bar on the left.
- The dimensions of $ p $ are shown as $ R \times 1 $, indicating that it is a vector with $ R $ elements.
- These inputs are processed by the weight matrix $ W $, which has $ R $ columns and 1 row (since this is a single neuron).
- A constant input of 1 is multiplied by the **bias** $ b $.
- The **net input** to the transfer function $ f $ is:

$$
n = Wp + b
$$

- The neuron output $ a $ is a scalar, as shown by:

$$
a = f(Wp + b)
$$

### Dimensions of Variables

- The dimensions of the variables are explicitly shown in this notation:
    - $ p $ is an $ R \times 1 $ vector,
    - $ W $ is a $ 1 \times R $ matrix,
    - $ n $ is a $ 1 \times 1 $ scalar, and
    - $ a $ is a $ 1 \times 1 $ scalar output.
  
This ensures that you can immediately recognize whether the variable is a scalar, vector, or matrix. No guessing is required!

## Network Architectures

Often, a single neuron—even with many inputs—may not be sufficient for solving complex problems. In such cases, we may need several neurons, say five or ten, operating in parallel. These neurons form what is known as a **layer**. This concept is discussed in more detail below.

### A Layer of Neurons


Figure 7. Layer of S Neurons

A **single-layer network** with $ S $ neurons is shown in **Figure 7**. Note that each of the $ R $ inputs is connected to each neuron in the layer, and the weight matrix $ W $ now has $ S $ rows.

#### Structure of the Layer

The structure of the layer includes:

- The **weight matrix** $ W $, which has dimensions $ S \times R $,
- **Summers** that compute the net input to each neuron,
- The **bias vector** $ b $, containing $ S $ biases (one for each neuron),
- **Transfer function boxes** $ f $, which define the output of each neuron,
- The **output vector** $ a $, which contains the outputs $ a_1, a_2, \dots, a_S $ of the neurons in the layer.

Each input element $ p_i $ is connected to all neurons through the weight matrix $ W $, and each neuron has its own bias $ b_i $, net input $ n_i $, transfer function $ f $, and output $ a_i $. Taken together, the outputs form the output vector $ a $, and the system can be described by:

$$
a = f(Wp + b)
$$

### Dimensions

In this model:

- $ p $ is an $ R \times 1 $ input vector,
- $ W $ is an $ S \times R $ weight matrix,
- $ b $ is an $ S \times 1 $ bias vector, and
- $ a $ is the $ S \times 1 $ output vector.

### Inputs and Layers

Some authors refer to the inputs as another layer; however, we do not do so here. The number of inputs $ R $ to a layer does not need to equal the number of neurons $ S $ in the layer (i.e., $ R \neq S $).

### Transfer Functions in a Layer

You might wonder whether all neurons in a layer must have the same transfer function. The answer is **no**. You can define a single (composite) layer of neurons with **different transfer functions** by combining multiple networks.



## Layer of S Neurons with Abbreviated Notation

In a network with multiple neurons, different networks can share the same inputs, and each network can generate some of the outputs. The **input vector elements** enter the network through the **weight matrix** $ W $, which is structured as follows:

$$
W = 
\begin{bmatrix}
w_{1,1} & w_{1,2} & \dots & w_{1,R} \\
w_{2,1} & w_{2,2} & \dots & w_{2,R} \\
\vdots  & \vdots  & \ddots & \vdots \\
w_{S,1} & w_{S,2} & \dots & w_{S,R}
\end{bmatrix}
\tag{2.6}
$$



### Indexing in the Weight Matrix

- The **row indices** of the elements in matrix $ W $ represent the **destination neuron** for the weight.
- The **column indices** represent the **source of the input** connected to that weight.
  
For instance, the weight $ w_{3,2} $ refers to the connection to the **third neuron** from the **second input source**.

### Abbreviated Notation for S-Neuron, R-Input Network

The **S-neuron**, **R-input**, **one-layer network** can also be represented in **abbreviated notation**, as shown in **Figure 2.8**. This representation simplifies the network while maintaining the necessary details about the inputs, weights, biases, and outputs.

Figure 8. Layer of S Neurons, Abbreviated Notation

In this case:

- $ p $ is an **input vector** of length $ R $ (i.e., $ R \times 1 $),
- $ W $ is an $ S \times R $ **weight matrix**,
- $ n $ is an $ S \times 1 $ **net input vector**,
- $ b $ is an $ S \times 1 $ **bias vector**, and
- $ a $ is the $ S \times 1 $ **output vector**.

The equation for the output of the layer remains the same:

$$
a = f(Wp + b)
$$

### Summary of Variables

- $ p $: Input vector of size $ R \times 1 $,
- $ W $: Weight matrix of size $ S \times R $,
- $ b $: Bias vector of size $ S \times 1 $,
- $ a $: Output vector of size $ S \times 1 $.

Thus, the **layer of neurons** includes the weight matrix $ W $, summation and multiplication operations, bias vector $ b $, transfer function $ f $, and output vector $ a $.




## Multiple Layers of Neurons

Now consider a neural network with multiple layers. Each layer has its own:

- **Weight matrix** $ W $,
- **Bias vector** $ b $,
- **Net input vector** $ n $, and
- **Output vector** $ a $.

### Notation for Multiple Layers

To distinguish between these layers, we will use **superscripts** to identify the corresponding layer number. For instance:

- The weight matrix for the **first layer** is denoted as $ W^1 $,
- The weight matrix for the **second layer** is denoted as $ W^2 $, and so on.

This notation is used to describe the **three-layer network** shown in **Figure 9**.


#### Figure 9: Three-Layer Network

$$
\begin{aligned}
    &\text{First Layer:} \\
    &a^1 = f^1(W^1 p + b^1) \\
    &\text{Second Layer:} \\
    &a^2 = f^2(W^2 a^1 + b^2) \\
    &\text{Third Layer (Output Layer):} \\
    &a^3 = f^3(W^3 a^2 + b^3)
\end{aligned}
$$

The overall output of the network can be written as:

$$
a^3 = f^3\left(W^3 f^2\left(W^2 f^1(W^1 p + b^1) + b^2\right) + b^3\right)
$$

### Explanation of the Layers

1. **First Layer**: This layer takes $ R $ inputs and has $ S_1 $ neurons. The output of this layer is $ a^1 $, which is used as the input for the second layer.
2. **Second Layer**: This layer takes the output of the first layer $ a^1 $ as its input and has $ S_2 $ neurons. The output of the second layer is $ a^2 $, which is passed to the third layer.
3. **Third Layer (Output Layer)**: This is the final layer, which produces the network's overall output, $ a^3 $.

### Notation for Layer Connections

- $ p $ is the input vector to the first layer.
- $ W^1 $, $ W^2 $, and $ W^3 $ are the weight matrices for the first, second, and third layers, respectively.
- $ b^1 $, $ b^2 $, and $ b^3 $ are the bias vectors for the respective layers.

### Layer Configuration

As shown in **Figure 2.9**, the network can have **different numbers of neurons in different layers**. The outputs of the first and second layers are used as inputs for the subsequent layers. 

- Layer 2 can be viewed as a one-layer network with $ R = S_1 $ inputs and $ S_2 $ neurons. The input to layer 2 is $ a^1 $, and the output is $ a^2 $.
  
A layer whose output is the **network output** is referred to as an **output layer**, while the intermediate layers are called **hidden layers**. In this case, the network has:

- Two **hidden layers** (layers 1 and 2),
- One **output layer** (layer 3).

### Abbreviated Notation for Three-Layer Network

The same three-layer network can be expressed using **abbreviated notation** as shown in **Figure 10**.


Figure 10. Three-Layer Network, Abbreviated Notation

Multilayer networks are more powerful than single-layer networks. For in- stance, a two-layer network having a sigmoid first layer and a linear second layer can be trained to approximate most functions arbitrarily well. Single- layer networks cannot do this. At this point the number of choices to be made in specifying a network may look overwhelming, so let us consider this topic. The problem is not as bad as it looks. First, recall that the number of inputs to the network and the number of outputs from the network are defined by external problem spec- ifications. So if there are four external variables to be used as inputs, there are four inputs to the network. Similarly, if there are to be seven outputs from the network, there must be seven neurons in the output layer. Finally, the desired characteristics of the output signal also help to select the trans- fer function for the output layer. If an output is to be either – 1 or 1 , then a symmetrical hard limit transfer function should be used. Thus, the archi- tecture of a single-layer network is almost completely determined by prob- lem specifications, including the specific number of inputs and outputs and the particular output signal characteristic. Now, what if we have more than two layers? Here the external problem does not tell you directly the number of neurons required in the hidden lay- ers. In fact, there are few problems for which one can predict the optimal number of neurons needed in a hidden layer. This problem is an active area of research. We will develop some feeling on this matter as we proceed to Chapter 11, Backpropagation. As for the number of layers, most practical neural networks have just two or three layers. Four or more layers are used rarely. We should say something about the use of biases. One can choose neurons with or without biases. The bias gives the network an extra variable, and so you might expect that networks with biases would be more powerful

```python
import numpy as np

class MultiLayerNeuralNetwork:
    def __init__(self, input_size, hidden_sizes, output_size):
        # Initialize the network architecture
        self.layers = []
        self.biases = []
        self.weights = []

        # Create hidden layers
        previous_size = input_size
        for size in hidden_sizes:
            self.weights.append(np.random.randn(previous_size, size) * 0.01)  # Initialize weights
            self.biases.append(np.zeros((1, size)))  # Initialize biases
            previous_size = size

        # Create output layer
        self.weights.append(np.random.randn(previous_size, output_size) * 0.01)
        self.biases.append(np.zeros((1, output_size)))

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def sigmoid_derivative(self, z):
        return z * (1 - z)

    def forward(self, X):
        # Forward propagation
        a = X
        self.layer_outputs = [X]  # Store the inputs

        for w, b in zip(self.weights, self.biases):
            z = np.dot(a, w) + b  # Weighted sum + bias
            a = self.sigmoid(z)    # Activation function
            self.layer_outputs.append(a)  # Store activations for later use

        return a  # Final output

    def backward(self, X, y, learning_rate=0.01):
        # Backward propagation
        m = y.shape[0]
        output_layer_index = len(self.layers)  # Last layer index
        output = self.layer_outputs[-1]

        # Calculate output layer error
        error = output - y
        dZ = error * self.sigmoid_derivative(output)

        # Update weights and biases for the output layer
        self.weights[-1] -= (1 / m) * np.dot(self.layer_outputs[-2].T, dZ) * learning_rate
        self.biases[-1] -= (1 / m) * np.sum(dZ, axis=0, keepdims=True) * learning_rate

        # Backpropagate the error to hidden layers
        for i in reversed(range(output_layer_index - 1)):
            dZ = np.dot(dZ, self.weights[i + 1].T) * self.sigmoid_derivative(self.layer_outputs[i + 1])
            self.weights[i] -= (1 / m) * np.dot(self.layer_outputs[i].T, dZ) * learning_rate
            self.biases[i] -= (1 / m) * np.sum(dZ, axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs=1000, learning_rate=0.01):
        for epoch in range(epochs):
            self.forward(X)  # Forward pass
            self.backward(X, y, learning_rate)  # Backward pass

    def predict(self, X):
        return self.forward(X)

# Example usage
if __name__ == "__main__":
    # XOR problem data
    X = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
    y = np.array([[0],
                  [1],
                  [1],
                  [0]])

    # Initialize and train the neural network
    nn = MultiLayerNeuralNetwork(input_size=2, hidden_sizes=[2], output_size=1)
    nn.train(X, y, epochs=10000, learning_rate=0.1)

    # Predictions
    print("Predictions:")
    print(nn.predict(X))
```

```python
import numpy as np
import matplotlib.pyplot as plt

class MultiLayerNeuralNetwork:
    def __init__(self, input_size, hidden_sizes, output_size):
        self.layers = []
        self.biases = []
        self.weights = []
        self.losses = []  # To track loss for plotting

        previous_size = input_size
        for size in hidden_sizes:
            self.weights.append(np.random.randn(previous_size, size) * 0.01)
            self.biases.append(np.zeros((1, size)))
            previous_size = size

        self.weights.append(np.random.randn(previous_size, output_size) * 0.01)
        self.biases.append(np.zeros((1, output_size)))

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def sigmoid_derivative(self, z):
        return z * (1 - z)

    def forward(self, X):
        a = X
        self.layer_outputs = [X]

        for w, b in zip(self.weights, self.biases):
            z = np.dot(a, w) + b
            a = self.sigmoid(z)
            self.layer_outputs.append(a)

        return a

    def compute_loss(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)  # Mean Squared Error

    def backward(self, X, y, learning_rate=0.01):
        m = y.shape[0]
        output = self.layer_outputs[-1]
        loss = self.compute_loss(y, output)
        self.losses.append(loss)  # Store the loss for plotting

        error = output - y
        dZ = error * self.sigmoid_derivative(output)

        self.weights[-1] -= (1 / m) * np.dot(self.layer_outputs[-2].T, dZ) * learning_rate
        self.biases[-1] -= (1 / m) * np.sum(dZ, axis=0, keepdims=True) * learning_rate

        for i in reversed(range(len(self.weights) - 1)):
            dZ = np.dot(dZ, self.weights[i + 1].T) * self.sigmoid_derivative(self.layer_outputs[i + 1])
            self.weights[i] -= (1 / m) * np.dot(self.layer_outputs[i].T, dZ) * learning_rate
            self.biases[i] -= (1 / m) * np.sum(dZ, axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs=1000, learning_rate=0.01):
        for epoch in range(epochs):
            self.forward(X)
            self.backward(X, y, learning_rate)

    def predict(self, X):
        return self.forward(X)

    def plot_loss(self):
        plt.figure(figsize=(8, 5))
        plt.plot(self.losses, label='Loss')
        plt.title('Training Loss Over Epochs')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid()
        plt.show()

    def plot_decision_boundary(self, X, y):
        h = .02  # step size in the mesh
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        Z = self.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z > 0.5  # Classify output
        Z = Z.reshape(xx.shape)

        plt.figure(figsize=(8, 5))
        plt.contourf(xx, yy, Z, alpha=0.8)
        plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), edgecolors='k', marker='o')
        plt.title('Decision Boundary')
        plt.xlabel('Input Feature 1')
        plt.ylabel('Input Feature 2')
        plt.grid()
        plt.show()

# Example usage
if __name__ == "__main__":
    # XOR problem data
    X = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
    y = np.array([[0],
                  [1],
                  [1],
                  [0]])

    # Initialize and train the neural network
    nn = MultiLayerNeuralNetwork(input_size=2, hidden_sizes=[2], output_size=1)
    nn.train(X, y, epochs=10000, learning_rate=0.1)

    # Predictions
    predictions = nn.predict(X)
    print("Predictions:")
    print(predictions)

    # Plot loss over epochs
    nn.plot_loss()

    # Plot decision boundary
    nn.plot_decision_boundary(X, y)
```
