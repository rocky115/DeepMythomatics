## Radial Basis Networks


## Introduction
Multilayer networks, as discussed in Chapters 11 and 12, are a type of neural network structure used for function approximation and pattern recognition. These networks utilize:

- **Sigmoid transfer functions** in hidden layers.
- **Linear transfer functions** in the output layer.

As established in Chapter 11, such networks are considered **universal function approximators**. In this chapter, we will explore another type of universal approximation network: the **Radial Basis Function (RBF) Network**.

## Overview of Radial Basis Function Networks
The RBF network can be applied to many of the same applications as multilayer networks. We will discuss the following topics:

1. **Universal approximation capabilities** of RBF networks.
2. **Training techniques** for these networks.
   - Gradient-based algorithms (similar to those discussed in Chapters 11 and 12).
   - A two-stage process for computing first-layer weights independently from second-layer weights.
   - Incremental building of the network, adding one neuron at a time.

### Theory and Examples
The radial basis function network is closely related to the multilayer perceptron network. It serves as a universal approximator suitable for:

- **Function approximation**
- **Pattern recognition**

The original work on radial basis functions was performed by Powell and others in the 1980s \cite{Powe87}. They utilized radial basis functions for **exact interpolation** in a multidimensional space, meaning the function created had to pass exactly through all targets in the training set.

However, in practice, we often deal with **noisy data**. As discussed in Chapter 13, exact interpolation can lead to overfitting when the training data is noisy. Thus, our focus will be on using radial basis functions to provide **robust approximations** to unknown functions based on limited and noisy measurements.

Broomhead and Lowe \cite{BrLo88} were the pioneers of the RBF neural network model, emphasizing the generation of a **smooth interpolating function**. Instead of forcing the network response to match target outputs exactly, the aim is to develop networks that generalize well to new situations.

## Demonstration of Capabilities
In this section, we will demonstrate the capabilities of radial basis function neural networks. 

## Training Procedures
The following sections will describe various procedures for training RBF networks.

\begin{equation}
f(x) = \sum_{i=1}^{N} w_i \phi(\| x - c_i \|) 
\end{equation}

Where:
- $ f(x) $ is the output function.
- $ w_i $ are the weights associated with each radial basis function.
- $ \phi $ is the radial basis function, which can be Gaussian or another suitable form.
- $ c_i $ are the centers of the radial basis functions.

## Conclusion
Radial basis function networks offer an alternative approach to multilayer networks for function approximation and pattern recognition. Their ability to generalize well from noisy data makes them a valuable tool in various applications.

## Radial Basis Function Network (RBF)

## Overview of Radial Basis Function Networks
The Radial Basis Function (RBF) network is a two-layer network that operates differently from a traditional two-layer perceptron network. There are two major distinctions:

1. **Distance Calculation**: In the first layer of the RBF network, instead of performing an inner product operation between the weights and the input (matrix multiplication), we calculate the distance between the input vector and the rows of the weight matrix. This approach is similar to the Learning Vector Quantization (LVQ) network.

2. **Bias Application**: Instead of adding a bias term, we multiply by the bias. The net input for neuron $ i $ in the first layer is calculated as follows:

$$
n_i = p - i \cdot w \cdot b_i \tag{17.1}
$$

### Key Concepts
- Each row of the weight matrix acts as a **center point**, where the net input value will be zero.
- The **bias** performs a scaling operation on the transfer (basis) function, allowing it to stretch or compress. 

**Note**: Most literature on RBF networks uses the terms **standard deviation**, **variance**, or **spread constant** instead of bias. This text uses the term "bias" for consistency with other networks.

### Gaussian Transfer Function
When a Gaussian transfer function is employed, the relationship between the bias $ b $ and the standard deviation $ \sigma $ is defined as follows:

$$
b = \frac{1}{\sigma^2} 
$$

The Gaussian function, commonly used in the neural network community, is defined as:

$$
a = e^{-n^2} \tag{17.2}
$$

### Properties of the Gaussian Function
A key property of the Gaussian function is its **locality**; the output approaches zero as the distance from the center point increases. This contrasts with global sigmoid functions, whose outputs remain close to 1 as the net input goes to infinity.

### Second Layer of the RBF Network
The second layer of the RBF network is a standard linear layer, represented as:

$$
a_2 = W_2 a_1 + b_2 \tag{17.3}
$$

### Complete RBF Network Structure
The complete RBF network structure consists of:

- **Inputs**
- **Radial Basis Layer**
- **Linear Layer**

The operation can be summarized as follows:

$$
a_i = \text{radbas}(\| w_i - p \| b_i) 
$$

$$
a = W a + b 
$$

### Flexibility of the RBF Network
This example demonstrates the flexibility of the RBF network for function approximation. As with the MLP, it is evident that if we have enough neurons in the first layer of the RBF network, we can approximate virtually any function of interest, and \cite{PaSa93} provides a mathematical proof of this capability. 

However, although both MLP and RBF networks are universal approximators, they perform their approximations in different ways. For the RBF network, each transfer function is only active over a small region of the input space  the response is local. If the input moves far from a given center, the output of the corresponding neuron will be close to zero. 

This has implications for the design of RBF networks. We must have centers adequately distributed throughout the range of the network inputs, and we must select biases in such a way that all of the basis functions overlap significantly (recall that the biases change the width of each basis function). 

We will discuss these design considerations in more detail in later sections.

```python
import matplotlib.pyplot as plt
import math

class RBFNetwork:
    def __init__(self, centers, biases, weights):
        self.centers = centers  # List of center points for each RBF neuron
        self.biases = biases      # List of biases for each RBF neuron
        self.weights = weights    # Weights for the output layer

    def gaussian(self, distance, bias):
        """Calculate the Gaussian basis function."""
        return math.exp(-bias * distance ** 2)

    def rbf(self, x):
        """Calculate the output of the RBF layer."""
        rbf_outputs = []
        for center, bias in zip(self.centers, self.biases):
            distance = abs(x - center)
            rbf_outputs.append(self.gaussian(distance, bias))
        return rbf_outputs

    def predict(self, x):
        """Calculate the network output for a given input x."""
        rbf_outputs = self.rbf(x)
        # Compute the output layer as a weighted sum of RBF outputs
        output = sum(w * o for w, o in zip(self.weights, rbf_outputs))
        return output

def plot_rbf_network(network, input_range):
    """Plot the RBF network response over a specified input range."""
    x_values = []
    y_values = []

    for x in input_range:
        y = network.predict(x)
        x_values.append(x)
        y_values.append(y)

    plt.plot(x_values, y_values, label='RBF Network Output')
    plt.title('RBF Network Function Approximation')
    plt.xlabel('Input (p)')
    plt.ylabel('Output (a)')
    plt.grid()
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.legend()
    plt.show()

# Define the RBF Network parameters
centers = [1, -1]      # Centers of the RBF neurons
biases = [2, 2]        # Biases for the RBF neurons (1/b)
weights = [1, 1]       # Weights for the output layer

# Create an instance of the RBF Network
rbf_network = RBFNetwork(centers, biases, weights)

# Define the input range for plotting
input_range = [x * 0.1 for x in range(-20, 21)]  # Input from -2 to 2 in steps of 0.1

# Plot the RBF Network response
plot_rbf_network(rbf_network, input_range)
```

# Pattern Classification with RBF Networks

To illustrate the capabilities of the Radial Basis Function (RBF) network for pattern classification, we will consider the classic exclusive-or (XOR) problem. The categories for the XOR gate are:

$$
\begin{aligned}
& \text{Category 1: } \\
& p_1 = -1, \quad p_2 = -1, \quad p_3 = 1 \\
& \text{Category 2: } \\
& p_1 = 4, \quad p_2 = 1, \quad p_3 = -1 \\
\end{aligned}
$$

The problem is illustrated graphically below. Because the two categories are not linearly separable, a single-layer network cannot perform the classification. However, RBF networks can classify these patterns.

In this example, the idea will be to have the network produce outputs greater than zero when the input is near patterns \( p_2 \) or \( p_3 \), and outputs less than zero for all other inputs. Note that the procedures we will use to design this example network are not suitable for complex problems, but they will help us illustrate the capabilities of the RBF network.

From the problem statement, we know that the network will need to have two inputs and one output. For simplicity, we will use only two neurons in the first layer (two basis functions), as this will be sufficient to solve the XOR problem.

As discussed earlier, the rows of the first-layer weight matrix will create centers for the two basis functions. We will choose the centers to be equal to the patterns \( p_2 \) and \( p_3 \). By centering a basis function at each pattern, we can produce maximum network outputs there. The first layer weight matrix is then:

$$
W = 
\begin{bmatrix}
p_2^T \\
p_3^T
\end{bmatrix}
=
\begin{bmatrix}
-1 & 1 \\
1 & -1
\end{bmatrix}
$$

The choice of the bias in the first layer depends on the width that we want for each basis function. For this problem, we would like the network function to have two distinct peaks at $ p_2 $ and $ p_3 $. Therefore, we dont want the basis functions to overlap too much.

The centers of the basis functions are each a distance of 2 from the origin. We want the basis function to drop significantly from its peak at that distance. If we use a bias of 1, we would get the following reduction in that distance:

$$
\begin{aligned}
2 \cdot 2 - n &= - (1 \cdot 2) \\
&= -2 \\
a &= e^{-2} \\
&= 0.1353
\end{aligned}
$$

Therefore, each basis function will have a peak of 1 at the centers and will drop to approximately 0.1353 at the origin. This will work for our problem, so we select the first layer bias vector to be:

$$
b = 
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

The original basis function response ranges from 0 to 1 (see Figure.1). We want the output to be negative for inputs much different than $ p_2 $ and $ p_3 $, so we will use a bias of -1 for the second layer, and we will use a value of 2 for the second layer weights to bring the peaks back up to 1. The second layer weights and biases then become:

$$
W = 
\begin{bmatrix}
2 \\
2
\end{bmatrix}, \quad b = -1
$$

For the network parameter values given above, the network response is shown in Figure 17.5. This figure also shows where the surface intersects the plane at \( a = 0 \), which is where the decision takes place. This is also indicated by the contours shown underneath the surface.

These are the function contours where $ a = 0 $. They are almost circles that surround the $ p_2 $ and $ p_3 $ vectors. This means that the network output will be greater than 0 only when the input vector is near the $ p_2 $ and $ p_3 $ vectors.

Figure.6 illustrates more clearly the decision boundaries. Whenever the input falls in the blue regions, the output of the network will be greater than zero. Whenever the network input falls outside the blue regions, the network output will be less than zero.

This network, therefore, classifies the patterns correctly. It is not the best solution, in the sense that it does not always assign input patterns to the closest prototype vector, unlike the MLP solution shown in Figure.2.

You will notice that the decision regions for this RBF network are circles, unlike the linear boundaries that we see in single-layer perceptrons. The MLP can put linear boundaries together to create arbitrary decision regions. The RBF network can put circular boundaries together to create arbitrary decision regions. In this problem, the linear boundaries are more efficient.

Of course, when many neurons are used, and the centers are close together, the elementary RBF boundaries are no longer purely circular, and the elementary MLP boundaries are no longer purely linear. However, associating circular boundaries with RBF networks and linear boundaries with MLP networks can be helpful in understanding their operation as pattern classifiers.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the RBF network class
class RBFNetwork:
    def __init__(self, centers, bias):
        self.centers = centers
        self.bias = bias
    
    def gaussian(self, x, center, bias):
        # Gaussian function for RBF
        return np.exp(-bias * np.linalg.norm(x - center)**2)

    def forward(self, x):
        # Calculate the output of the RBF network
        rbf_output = np.array([self.gaussian(x, center, self.bias) for center in self.centers])
        # Linear combination (output layer)
        output = np.dot(np.array([2, 2]), rbf_output) - 1  # Weights for the output layer
        return output

# Define centers and bias for the RBF
centers = np.array([[-1, 1], [1, -1]])  # Centers for the RBFs
bias = 1  # Bias for the Gaussian functions

# Create the RBF network
rbf_network = RBFNetwork(centers, bias)

# Generate data points for XOR
x_values = np.linspace(-3, 3, 100)
y_values = np.linspace(-3, 3, 100)
X1, Y1 = np.meshgrid(x_values, y_values)
Z = np.zeros(X1.shape)

# Calculate the network output for each point in the grid
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        x = np.array([X1[i, j], Y1[i, j]])
        Z[i, j] = rbf_network.forward(x)

# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(X1, Y1, Z, levels=np.linspace(-2, 2, 50), cmap='RdBu', alpha=0.8)
plt.colorbar()
plt.scatter(centers[:, 0], centers[:, 1], color='black', s=100, label='Centers')
plt.title('RBF Network Decision Boundary for XOR Problem')
plt.xlabel('p1')
plt.ylabel('p2')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Define the Gaussian RBF function
def gaussian_rbf(distance, bias):
    return math.exp(-bias * (distance ** 2))

# Define the RBF network class
class RBFNetwork:
    def __init__(self, centers, bias, weights, bias_output):
        self.centers = centers          # Centers of the RBFs
        self.bias = bias                # Bias (spread) of the RBFs
        self.weights = weights          # Weights for the output layer
        self.bias_output = bias_output  # Bias for the output layer

    def forward(self, input):
        # Calculate the hidden layer outputs (RBF outputs)
        rbf_outputs = []
        for center in self.centers:
            distance = math.sqrt((input[0] - center[0]) ** 2 + (input[1] - center[1]) ** 2)
            rbf_outputs.append(gaussian_rbf(distance, self.bias))

        # Calculate the final output
        output = sum(w * rbf for w, rbf in zip(self.weights, rbf_outputs)) + self.bias_output
        return output

# Define XOR input patterns and corresponding outputs
X = [(-1, -1), (-1, 1), (1, -1), (1, 1)]  # Input patterns
y = [-1, 1, 1, -1]                        # Expected outputs

# Define RBF parameters
centers = [(-1, 1), (1, -1)]   # Centers for the RBFs (p2 and p3)
bias = 1                       # Spread of the RBFs
weights = [2, 2]              # Weights for the output layer
bias_output = -1              # Bias for the output layer

# Create the RBF network
rbf_network = RBFNetwork(centers, bias, weights, bias_output)

# Test the network on input patterns and collect results
results = []
for input_pattern in X:
    output = rbf_network.forward(input_pattern)
    results.append(output)

# Print the network outputs
for input_pattern, output in zip(X, results):
    print(f"Input: {input_pattern}, Output: {output}")

# Create a grid for visualization
grid_size = 100
x_values = [i * 4 / grid_size - 2 for i in range(grid_size)]
y_values = [i * 4 / grid_size - 2 for i in range(grid_size)]
Z = [[rbf_network.forward((x, y)) for x in x_values] for y in y_values]

# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(x_values, y_values, Z, levels=50, cmap='coolwarm', alpha=0.6)
plt.colorbar(label='Output')
plt.scatter([-1, 1], [-1, 1], c='red', label='Class 1 (p2, p3)', s=100)
plt.scatter([-1, 1], [1, -1], c='blue', label='Class 2', s=100)
plt.title('RBF Network Decision Boundary for XOR')
plt.xlabel('p1')
plt.ylabel('p2')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.show()
```

# Global vs. Local Representations in Neural Networks

## Global vs. Local Transfer Functions

The global transfer functions used by **Multi-Layer Perceptrons (MLPs)** and the local transfer functions employed by **Radial Basis Function (RBF)** networks each have their advantages and disadvantages:

- **MLP Networks:**
  - Create a **distributed representation** because all of the transfer functions overlap in their activity.
  - At any given input value, many sigmoid functions in the first layer will have significant outputs. They must sum or cancel in the second layer to produce the appropriate response at each point.

- **RBF Networks:**
  - Each basis function is only active over a small range of the input. For any given input, only a few basis functions will be active.
  
### Advantages and Disadvantages

| Aspect               | MLP Networks                           | RBF Networks                          |
|----------------------|---------------------------------------|---------------------------------------|
| **Neuron Count**     | Fewer neurons needed due to global coverage | More neurons may be required to cover the input space accurately |
| **Training Speed**   | Typically slower due to the complexity of optimization | Generally faster training with two-stage algorithms |
| **Overfitting**      | Less prone to overfitting due to fewer parameters | More likelihood of overfitting with more neurons |
| **Adaptability**     | Global representation may improve accuracy in certain regions | Local representation reduces risk of overfitting in unused regions |

## Training RBF Networks

### Training Approaches

Unlike the MLP network, which is almost always trained by gradient-based algorithms, the RBF network can be trained using various methods. 

- **Gradient-Based Algorithms:**
  - These can be used for fine-tuning after initial training but often suffer from local minima issues.

- **Two-Stage Algorithms:**
  - The two layers of the RBF network are treated separately.
  - The first layer weights and biases are selected before computing the second layer weights in one step using a linear least-squares algorithm.

### Common Two-Stage Training Methods

1. **Grid Pattern Centers:**
   - Centers arranged in a grid throughout the input range.
   - Fixed bias chosen for overlap, but this method is suboptimal.

2. **Random Selection:**
   - Centers randomly selected from training input vectors.
   - Ensures basis centers are placed in useful areas but may not be optimal.

3. **Clustering Methods:**
   - Use methods like the **Kohonen competitive layer** to cluster the input space for basis function centers.

4. **Orthogonal Least Squares:**
   - A method that starts with many possible centers and adds one at a time based on the reduction of sum squared error.

## Linear Least Squares for RBF Training

When the first layer weights and biases of the RBF network are fixed, the second layer weights can be computed using linear least squares. The biases can be computed using the following formula:

$$
b_j = \frac{1}{N} \sum_{i=1}^{N} \phi_j(x_i)
$$

where:
- $ b_j $ is the bias for the $ j^{th} $ basis function,
- $ N $ is the number of training samples,
- $ \phi_j(x_i) $ is the output of the $ j^{th} $ basis function for the $ i^{th} $ training sample.


## Linear Least Squares in RBF Networks

The bias for the $ i^{th} $ basis function can be defined as:

$$
b_i = \frac{1}{d_{\text{max}}}, \quad (17.9)
$$

where $ d_{\text{max}} $ is the maximum distance between neighboring centers. This ensures an appropriate degree of overlap between the basis functions, and using this method results in all biases having the same value.

### Training the Second Layer

Once the first layer parameters have been set, training the second layer weights and biases is equivalent to training a linear network. Consider the following training points:

$$
\{p_1, t_1\}, \{p_2, t_2\}, \ldots, \{p_Q, t_Q\}, \quad (17.10)
$$

where $ p_q $ is an input to the network and $ t_q $ is the corresponding target output. The output of the first layer for each input $ p_q $ in the training set can be computed as:

$$
n_{i,q} = p_q - iw b_i, \quad (17.11)
$$

The activation of the $ i^{th} $ basis function is given by:

$$
a_q = \text{radbas}(n_{q}), \quad (17.12)
$$

The training data set for the second layer becomes:

$$
\{ a_1, t_1 \}, \{ a_2, t_2 \}, \ldots, \{ a_Q, t_Q \} \quad (17.13)
$$

The response of the second layer is linear:

$$
a_2 = W a + b, \quad (17.14)
$$

To minimize the sum square error performance index over the training set, we define:

$$
F(x) = \sum_{q=1}^{Q} (t_q - a_q)^2, \quad (17.15)
$$

### Deriving the Solution

Assuming a scalar target, we lump all parameters we are adjusting, including the bias, into one vector:

$$
x = \begin{bmatrix} 1 \\ w \end{bmatrix}, \quad (17.16)
$$

Similarly, we include the bias input "1" as a component of the input vector:

$$
z_q = a_q \quad (17.17)
$$

Now the network output can be expressed as:

$$
a_q = (1 \; w)^T a_q + b, \quad (17.18)
$$

This can be written conveniently as:

$$
a_q = x^T z_q, \quad (17.19)
$$

The sum square error can be expressed as:

$$
F(x) = \sum_{q=1}^{Q} (e_q)^2 = \sum_{q=1}^{Q} (t_q - a_q)^2 = \sum_{q=1}^{Q} (t_q - x^T z_q). \quad (17.20)
$$

### Matrix Formulation

To express this in matrix form, we define the following matrices:

$$
U = \begin{bmatrix} t_1 \\ t_2 \\ \vdots \\ t_Q \end{bmatrix}, \quad z = \begin{bmatrix} z_1 \\ z_2 \\ \vdots \\ z_Q \end{bmatrix}, \quad e = \begin{bmatrix} e_1 \\ e_2 \\ \vdots \\ e_Q \end{bmatrix}. \quad (17.21)
$$

The error can now be written as:

$$
e = t - Ux, \quad (17.22)
$$

The performance index becomes:

$$
F(x) = (t - Ux)^T (t - Ux). \quad (17.23)
$$

### Regularization

If we use regularization to prevent overfitting, we obtain the following form for the performance index:


## Regularization in Linear Least Squares

The regularized least squares solution can be expressed as:

$$
\left[ U^T U + \rho I \right] x^* = U^T t, \quad (17.30)
$$

where:
- $ U $ is the design matrix,
- $ \rho $ is the regularization parameter, and
- $ I $ is the identity matrix.

If the Hessian matrix is positive definite, there will be a unique stationary point, which will be a strong minimum:

$$
x^* = \left[ U^T U + \rho I \right]^{-1} U^T t. \quad (17.31)
$$

### Demonstration of the Procedure

Lets demonstrate this procedure with a simple problem.


```python
import matplotlib.pyplot as plt
import numpy as np

# Radial Basis Function (RBF)
def radbas(x):
    return np.exp(-np.power(x, 2))

# RBF Network Class
class RBFNetwork:
    def __init__(self, centers, bias, output_weights):
        self.centers = centers
        self.bias = bias
        self.output_weights = output_weights

    def _calculate_activation(self, x):
        # Calculate distances from each center
        distances = []
        for center in self.centers:
            distance = np.sqrt((x[0] - center[0]) ** 2 + (x[1] - center[1]) ** 2)
            distances.append(distance)

        # Apply radial basis function
        activations = [radbas(d - self.bias) for d in distances]
        return activations

    def predict(self, x):
        activations = self._calculate_activation(x)
        output = np.dot(self.output_weights, activations)
        return output

# Define the XOR patterns and target outputs
patterns = [(-1, -1), (1, -1), (-1, 1), (1, 1)]
targets = [-1, 1, 1, -1]

# Define RBF centers and parameters
centers = [(-1, 1), (1, -1)]
bias = 1
output_weights = [2, 2]  # weights for the output layer

# Create the RBF Network
rbf_network = RBFNetwork(centers, bias, output_weights)

# Plot decision boundaries
x_range = np.linspace(-3, 3, 100)
y_range = np.linspace(-3, 3, 100)
xx, yy = np.meshgrid(x_range, y_range)
Z = np.zeros(xx.shape)

# Compute output for each point in the grid
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        output = rbf_network.predict((xx[i, j], yy[i, j]))
        Z[i, j] = output

# Plotting
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, levels=np.linspace(-2, 2, 100), cmap='coolwarm', alpha=0.8)
plt.colorbar()
plt.scatter(*zip(*patterns), c='black', marker='o', label='Patterns')
plt.title('RBF Network Decision Boundaries for XOR Problem')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.axvline(0, color='grey', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```

## Example: RBF Network Using Least Squares Algorithm

To illustrate the least squares algorithm, we will use a Radial Basis Function (RBF) network with three neurons in the first layer to approximate the following function:

$$
g(p) = 1 + \sin\left(\frac{\pi p}{4}\right) \quad \text{for } -2 \leq p \leq 2.
$$

### Training Set

We will evaluate this function at six values of \( p \):

$$
p = \{-2, -1.2, -0.4, 0.4, 1.2, 2\}.
$$

This produces the targets:

$$
t = \{0, 0.19, 0.69, 1.3, 1.8, 2\}.
$$

### Basis Function Centers and Biases

We will choose the basis function centers to be spaced equally throughout the input range: \(-2, 0, \) and \(2\). For simplicity, we will choose the bias to be the reciprocal of the spacing between points. This produces the following first layer weights and biases:

$$
W = \begin{bmatrix}
0 & 1 & 1 \\
0.5 & 0.5 & 0.5
\end{bmatrix}, \quad b = \begin{bmatrix}
0.5 \\
0.5
\end{bmatrix}.
$$

### First Layer Output Computation

Next, we compute the output of the first layer using the equations:

$$
n_{i,q} = p_q - W_i b_i,
$$

$$
a_q = \text{radbas}(n_{i,q}).
$$

This produces the following output vectors:

$$
a = \begin{bmatrix}
1 & 0.852 & 0.527 & 0.237 & 0.077 & 0.018 \\
0.368 & 0.698 & 0.961 & 0.961 & 0.698 & 0.368 \\
0.018 & 0.077 & 0.237 & 0.527 & 0.852 & 1
\end{bmatrix}.
$$

### Matrices U and t

We can use the outputs \( a \) and the targets \( t \) to create the matrices \( U \) and \( t \):

$$
U = \begin{bmatrix}
1 & 0.852 & 0.527 & 0.237 & 0.077 & 0.018 \\
0.368 & 0.698 & 0.961 & 0.961 & 0.698 & 0.368 \\
0.018 & 0.077 & 0.237 & 0.527 & 0.852 & 1
\end{bmatrix},
$$

$$
t = \begin{bmatrix}
0 \\
0.19 \\
0.69 \\
1.3 \\
1.8 \\
2
\end{bmatrix}.
$$

### Solving for Weights and Biases in the Second Layer

Next, we solve for the weights and biases in the second layer using the equation:

$$
\mathbf{x}^* = \left[ U^T U + \rho I \right]^{-1} U^T t,
$$

with the regularization parameter $\rho $ set to zero.

The result is:

$$
\mathbf{x}^* = \begin{bmatrix}
-1.03 \\
0 \\
1.03 \\
1
\end{bmatrix}.
$$

### Final Weights and Biases

Thus, the second layer weights and bias are given by:

$$
W = \begin{bmatrix}
-1.03 & 0 & 1.03
\end{bmatrix}, \quad b = 1.
$$

### Illustration

Figure.8 RBF Response with Bias Too Large

Figure.7 illustrates the operation of this RBF network. The blue line represents the RBF approximation, and the circles represent the six data points. The dotted lines in the upper axis represent the individual basis functions scaled by the corresponding weights in the second layer (including the constant bias term). The sum of the dotted lines will produce the blue line.

In the small axis at the bottom, you can see the unscaled basis functions, which are the outputs of the first layer.


```python
import matplotlib.pyplot as plt
import math

# Define the function to approximate
def g(p):
    return 1 + math.sin(math.pi * p / 4)

# Generate training data
p_values = [-2, -1.2, -0.4, 0.4, 1.2, 2]
t_values = [g(p) for p in p_values]

# Define RBF parameters
centers = [-2, 0, 2]  # Basis function centers
num_centers = len(centers)
num_samples = len(p_values)

# Calculate biases (inverse of distance between centers)
d_max = centers[1] - centers[0]  # Assuming equal spacing
bias = 1 / d_max

# Calculate the output of the first layer
a = [[0] * num_samples for _ in range(num_centers)]

for i in range(num_centers):
    for q in range(num_samples):
        n_i_q = p_values[q] - centers[i] * bias
        a[i][q] = math.exp(-n_i_q ** 2)  # Radial basis function (Gaussian)

# Create U matrix
U = [[0] * (num_centers + 1) for _ in range(num_samples)]

for q in range(num_samples):
    for i in range(num_centers):
        U[q][i] = a[i][q]
    U[q][num_centers] = 1  # For bias term

# Prepare target vector
t = [[t_values[q]] for q in range(num_samples)]

# Calculate weights using least squares solution
def matrix_mult(A, B):
    result = [[0] * len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

def transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

def add_matrices(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def inverse_matrix(matrix):
    # Implementing Gaussian elimination for matrix inversion
    n = len(matrix)
    # Create augmented matrix
    augmented = [row[:] + ([1 if i == j else 0 for j in range(n)]) for i, row in enumerate(matrix)]
    
    for i in range(n):
        # Make the diagonal contain all 1's
        diag_val = augmented[i][i]
        for j in range(len(augmented[i])):
            augmented[i][j] /= diag_val
        
        # Make all rows below this one 0 in current column
        for j in range(i + 1, n):
            factor = augmented[j][i]
            for k in range(len(augmented[j])):
                augmented[j][k] -= factor * augmented[i][k]

    # Back substitution
    for i in range(n - 1, -1, -1):
        for j in range(i - 1, -1, -1):
            factor = augmented[j][i]
            for k in range(len(augmented[j])):
                augmented[j][k] -= factor * augmented[i][k]

    # Return only the inverse part
    return [row[n:] for row in augmented]

# Solve for weights and biases
U_transpose = transpose(U)
U_U_transpose = matrix_mult(U_transpose, U)

# Regularization parameter (set to 0 for this example)
rho = 0
identity_matrix = [[1 if i == j else 0 for j in range(num_centers + 1)] for i in range(num_centers + 1)]
U_U_transpose_plus_rhoI = add_matrices(U_U_transpose, identity_matrix)

# Calculate inverse of (U^T * U + rho * I)
inverse_part = inverse_matrix(U_U_transpose_plus_rhoI)

# Final weights calculation
U_transpose_t = matrix_mult(U_transpose, t)
W = matrix_mult(inverse_part, U_transpose_t)

# Prepare the approximation for plotting
p_plot = [x / 10.0 for x in range(-20, 21)]  # From -2 to 2
g_approx = []

for p in p_plot:
    a_plot = [math.exp(-(p - centers[i]) ** 2) for i in range(num_centers)]
    a_plot.append(1)  # Bias term
    g_value = 0
    for i in range(num_centers + 1):
        g_value += W[i][0] * a_plot[i]
    g_approx.append(g_value)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(p_plot, g_approx, label='RBF Approximation', color='blue')
plt.scatter(p_values, t_values, color='red', label='Training Points')
plt.title('RBF Network Approximation')
plt.xlabel('p')
plt.ylabel('g(p)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```

## Orthogonal Least Squares

In the previous section, we assumed that the weights and biases in the first layer were fixed. For example, the centers could be fixed on a grid or randomly selected from the input vectors in the training set. In this section, we consider a different approach for selecting the centers. We will assume that there exists a number of potential centers. These centers could include the entire set of input vectors in the training set, vectors chosen in a grid pattern, or vectors chosen by any other procedure one might think of. We will then select vectors one at a time from this set of potential centers until the network performance is satisfactory. We will build up the network one neuron at a time.

The basic idea behind this method comes from statistics, and it is called **subset selection** [Mill90]. The general objective of subset selection is to choose an appropriate subset of independent variables to provide the most efficient prediction of a target dependent variable. 

For example, suppose that we have \(Q\) independent variables, and we want to use them to predict our target dependent variable. We want to create the simplest predictor possible, so we want to use the minimum number of independent variables for the prediction. Which subset of the \(Q\) independent variables should we use?

The optimal approach, called an exhaustive search, tries all combinations of subsets and finds the smallest one that provides satisfactory performance. The number of distinct subsets is given by:

$$
\sum_{q=1}^{Q} \frac{Q!}{q!(Q - q)!} \tag{17.43}
$$

For \(Q = 10\), this number is 1023. If \(Q = 20\), the number is more than 1 million. Therefore, we need a less expensive strategy than the exhaustive search.

There are several suboptimal procedures that are not guaranteed to find the optimal subset, but they require significantly less computation. One procedure is called **forward selection**. This method begins with an empty model and then adds variables one at a time. At each stage, we add the independent variable that provides the largest reduction in squared error. We stop adding variables when the performance is adequate.

Another approach, called **backward elimination**, starts with all independent variables selected for the model. At each stage, we eliminate the variable that would cause the least increase in the squared error. The process continues until the performance is inadequate. There are other approaches which combine forward selection and backward elimination so that variables can be added and deleted at each iteration.

Any of the standard subset selection techniques can be used for selecting RBF centers. For purposes of illustration, we will consider one specific form of forward selection, called **orthogonal least squares** [ChCo91]. Its main feature is that it efficiently calculates the error reduction provided by the addition of each potential center to the RBF network.

To develop the orthogonal least squares algorithm, we begin with the equation:

$$
t = Ux + e \tag{17.44}
$$

We will use our standard notation for matrix rows and columns to individually identify both the rows and the columns of the matrix \(U\):

$$
U = \begin{bmatrix}
u_1^T \\
u_2^T \\
\vdots \\
u_Q^T
\end{bmatrix} = \begin{bmatrix}
z_1^T \\
z_2^T \\
\vdots \\
z_Q^T
\end{bmatrix} \tag{17.45}
$$

Here, each row of the matrix \(U\) represents the output of layer 1 of the RBF network for one input vector from the training set. There will be a column of the matrix \(U\) for each neuron (basis function) in layer 1 plus the bias term (i.e., $n = S + 1$). 

Note that for the OLS algorithm, the potential centers for the basis functions are often chosen to be all of the input vectors in the training set. In this case, $n$ will equal $Q + 1$, since the constant "1" for the bias term is included in $z$.

Equation (17.44) is in the form of a standard linear regression model. The matrix \(U\) is called the **regression matrix**, and the columns of \(U\) are called the **regressor vectors**.

The objective of OLS is to determine how many columns of \(U\) (numbers of neurons or basis functions) should be used. The first step is to calculate how much each potential column would reduce the squared error. 

The problem is that the columns are generally correlated with each other, making it difficult to determine how much each individual column would reduce the error. For this reason, we need to first orthogonalize the columns. Orthogonalizing the columns means that we can decompose \(U\) as follows:

$$
U = MR \tag{17.46}
$$

where \(R\) is an upper triangular matrix with ones on the diagonal:

$$
R = \begin{bmatrix}
1 & r_{1,2} & r_{1,3} & \cdots & r_{1,n} \\
0 & 1 & r_{2,3} & \cdots & r_{2,n} \\
0 & 0 & 1 & \cdots & r_{3,n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \cdots & 1
\end{bmatrix} \tag{17.47}
$$

and \(M\) is a matrix with orthogonal columns $m_i$. This means that $M$ has the following properties:

$$
M^T M = V = \begin{bmatrix}
v_{1,1} & 0 & \cdots & 0 \\
0 & v_{2,2} & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & v_{n,n}
\end{bmatrix} \tag{17.48}
$$

Now Equation (17.44) can be rewritten as:

$$
t = MRx + e = Mh + e \tag{17.49}
$$

where 

$$
h = Rx \tag{17.50}
$$

The least squares solution for Equation (17.49) is given by:

$$
h^* = (M^T M)^{-1} M^T t = V^{-1} M^T t \tag{17.51}
$$

Since $V$ is diagonal, the elements of $h^*$ can be computed as:

$$
h_i^* = \frac{m_i^T t}{v_{i,i}} \tag{17.52}
$$

From $h^*$, we can compute $x^*$ using Equation (17.50). Since $R$ is upper-triangular, Equation (17.50) can be solved by back-substitution and does not require a matrix inversion.

There are a number of ways to obtain the orthogonal vectors $m_i$, but we will use the **Gram-Schmidt orthogonalization** process of Equation (5.20), starting with the original columns of $U$:

$$
m_1 = u_1 \tag{17.53}
$$


## Orthogonal Least Squares (Continued)

The orthogonalization process allows us to express the residuals in terms of the basis vectors. We have:

$$
k^{-1} m_k = u_k - \sum_{i=1}^{k-1} r_{i,k} m_i \tag{17.54}
$$

where 

$$
r_{i,k} = \frac{m_i^T u_k}{m_i^T m_i}, \quad i = 1, \ldots, k - 1 \tag{17.55}
$$

Now, lets see how orthogonalizing the columns of \(U\) enables us to efficiently calculate the squared error contribution of each basis vector. Using Equation (17.49), the total sum square value of the targets is given by:

$$
t^T t = \left[ Mh + e \right]^T \left[ Mh + e \right] = h^T M^T Mh + e^T Mh + h^T M e + e^T e \tag{17.56}
$$

Consider the second term in the sum:

$$
e^T Mh = \left[ t - Mh \right]^T Mh = t^T Mh - h^T M^T Mh \tag{17.57}
$$

If we use the optimal $h^*$ from Equation (17.51), we find:

$$
e^T Mh^* = t^T Mh^* - t^T M V M h^* = t^T Mh^* - t^T Mh^* = 0 \tag{17.58}
$$

Therefore, the total sum square value from Equation (17.56) becomes:

$$
t^T t = h^T M^T Mh + e^T e = h^T V h + e^T e = \sum_{i=1}^{n} h_i m_i^T m_i + e^T e \tag{17.59}
$$

The first term on the right of Equation (17.59) is the contribution to the sum squared value explained by the regressors, and the second term is the remaining sum squared value that is not explained by the regressors. Therefore, regressor (basis function) \(i\) contributes:

$$
h_i^2 m_i^T m_i \tag{17.60}
$$

to the squared value. This also represents how much the squared error can be reduced by including the corresponding basis function in the network.

We will use this number, after normalizing by the total squared value, to determine the next basis function to include at each iteration:

$$
o_i = \frac{h_i^2 m_i^T m_i}{t^T t} \tag{17.61}
$$

This number always falls between zero and one. 

Now lets put all these ideas together into an algorithm for selecting centers.


```python
import math

# Define the Radial Basis Function
def radbas(x, width=1.0):
    """Radial Basis Function (Gaussian) with adjustable width."""
    return math.exp(-x**2 / (2 * width**2))

# Define the Gram-Schmidt orthogonalization process
def gram_schmidt(U, epsilon=1e-10):
    """Orthogonalize the columns of matrix U using Gram-Schmidt."""
    m, n = len(U), len(U[0])  # m: number of samples, n: number of features
    M = [[0] * n for _ in range(m)]  # Orthogonal matrix
    R = [[0] * n for _ in range(n)]  # Upper triangular matrix

    for k in range(n):
        # Copy the current column of U to M
        for i in range(m):
            M[i][k] = U[i][k]
        
        for j in range(k):
            # Calculate R[j][k] and adjust M
            R[j][k] = sum(M[i][j] * U[i][k] for i in range(m))
            for i in range(m):
                M[i][k] -= R[j][k] * M[i][j]
        
        # Normalize M[k]
        norm = math.sqrt(sum(M[i][k] ** 2 for i in range(m)))
        if norm < epsilon:
            raise ValueError("Zero vector encountered in Gram-Schmidt process.")
        for i in range(m):
            M[i][k] /= norm
        R[k][k] = norm

    return M, R

# Implement Orthogonal Least Squares
def orthogonal_least_squares(X, T, potential_centers, width=1.0):
    """Perform Orthogonal Least Squares to select RBF centers."""
    selected_centers = []
    residuals = T[:]  # Start with initial residuals equal to targets
    total_squared = sum(t ** 2 for t in T)  # Total sum of squares

    # Main loop to select centers
    while True:
        contributions = []
        
        for center in potential_centers:
            # Compute the output of the RBF network for the current center
            U = [[radbas((x - center), width) for x in X]]  # Radial basis outputs
            
            # Orthogonalize the columns of U
            try:
                M, R = gram_schmidt(U)
            except ValueError:
                # Skip this center if orthogonalization fails
                continue
            
            # Calculate the optimal weights
            h_star = [(sum(M[i][0] * T[j] for j in range(len(T))) / R[i][i]) for i in range(len(M))]
            
            # Calculate the contribution to the squared error reduction
            contribution = sum(h_i**2 * (M[i][0]**2) for i, h_i in enumerate(h_star))
            contributions.append(contribution / total_squared)

        if not contributions:
            break  # No valid contributions, end loop

        # Find the center that provides the maximum contribution
        max_contribution_index = contributions.index(max(contributions))
        selected_centers.append(potential_centers[max_contribution_index])
        
        # Remove the selected center from potential centers
        potential_centers.pop(max_contribution_index)
        
        # Check for stopping condition (could be based on contributions)
        if max(contributions) < 0.01:  # Example threshold
            break
            
    return selected_centers

# Example usage
X = [-2, -1.2, -0.4, 0.4, 1.2, 2]  # Input values
T = [0, 0.19, 0.69, 1.3, 1.8, 2]   # Target values
potential_centers = X[:]  # Use all input values as potential centers

# Perform Orthogonal Least Squares
selected_centers = orthogonal_least_squares(X, T, potential_centers, width=1.0)
print("Selected RBF Centers:", selected_centers)
```

$$
\textbf{OLS Algorithm}
$$

To begin the algorithm, we start with all potential basis functions included in the regression matrix \( \mathbf{U} \). As explained below Eq. (17.45), if all input vectors in the training set are to be considered potential basis function centers, then the \( \mathbf{U} \) matrix will be \( Q \times (Q + 1) \). This matrix represents only potential basis functions, as we start with no basis functions included in the network.

### First Stage of OLS Algorithm:

For \( i = 1, \ldots, Q \):

$$
m_1^{(i)} = \mathbf{u}_i \tag{17.62}
$$

$$
h_1^{(i)} = \frac{m_1^{(i)T} \mathbf{t}}{m_1^{(i)T} m_1^{(i)}} \tag{17.63}
$$

$$
o_1^{(i)} = \frac{(h_1^{(i)})^2 m_1^{(i)T} m_1^{(i)}}{\mathbf{t}^T \mathbf{t}} \tag{17.64}
$$

We select the basis function that creates the largest reduction in error:

$$
o_1^{(i_1)} = \max \{o_1^{(i)}\} \tag{17.65}
$$

$$
m_1 = m_1^{(i_1)} = \mathbf{u}_{i_1} \tag{17.66}
$$

### Remaining Iterations of the Algorithm:

For iteration \( k \):

For \( i = 1, \ldots, Q \), \( i \neq i_1, \ldots, i \neq i_{k-1} \):

$$
r_{j,k}^{(i)} = \frac{m_j^T \mathbf{u}_i}{m_j^T m_j}, \quad j = 1, \ldots, k-1 \tag{17.67}
$$

$$
m_k^{(i)} = \mathbf{u}_i - \sum_{j=1}^{k-1} r_{j,k}^{(i)} m_j \tag{17.68}
$$

$$
h_k^{(i)} = \frac{m_k^{(i)T} \mathbf{t}}{m_k^{(i)T} m_k^{(i)}} \tag{17.69}
$$

$$
o_k^{(i)} = \frac{(h_k^{(i)})^2 m_k^{(i)T} m_k^{(i)}}{\mathbf{t}^T \mathbf{t}} \tag{17.70}
$$

Select the basis function that creates the largest reduction in error:

$$
o_k^{(i_k)} = \max \{o_k^{(i)}\} \tag{17.71}
$$

### Stopping Criterion:

The iterations continue until some stopping criterion is met. One option is:

$$
\sum_{j=1}^{k-1} o_j < \delta \tag{17.74}
$$

where \( \delta \) is some small number. If \( \delta \) is too small, overfitting may occur as the network becomes too complex.

Alternatively, a validation set can be used, and the process stops when the error on the validation set increases.

### Weight Computation After Convergence:

Once the algorithm converges, the original weights \( \mathbf{x} \) can be computed from the transformed weights \( \mathbf{h} \) by back-substitution using:

$$
x_n = h_n, \quad x_k = h_k - \sum_{j=k+1}^n r_{j,k} x_j \tag{17.75}
$$

where $ n \) is the final number of adjustable parameters (weights and biases) in the second layer.
