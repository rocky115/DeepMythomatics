## Recurrent Networks and Nonlinear Dynamics

## Overview

Recurrent neural networks (RNNs) are characterized by feedback connections, allowing outputs to influence future inputs. This feedback mechanism enables RNNs to recognize and recall **temporal** as well as **spatial** patterns, making them more powerful than feedforward networks. However, this also introduces complexity in their behavior. Unlike feedforward networks, where output remains constant for a fixed input, the output of recurrent networks varies over time.

### Types of Recurrent Networks
- **Hamming and Hopfield Networks**: Introduced in Chapter 3, these networks utilize recurrent connections for pattern storage and recall.
- **Grossberg Networks**: Discussed in Chapters 18 and 19, these networks also incorporate recurrent connections to perform complex pattern recognition tasks.

## Temporal Dynamics of Recurrent Networks

In recurrent networks, the output depends on:
1. The network's **current state**.
2. The **initial output state** and **input over time**.

This makes the behavior of recurrent networks sensitive to initial conditions. Possible behaviors include:
- **Convergence**: Stabilizing to a fixed output.
- **Oscillation**: Repeating output patterns.
- **Divergence**: Output escalating towards infinity.
- **Chaos**: Unpredictable, non-repeating patterns.

### Nonlinear Recurrent Networks

To study the long-term behavior of RNNs, we describe them using nonlinear differential equations. These equations define how the network's output changes over time based on the current state and input.

A general form of these equations is:

$$
\frac{d a(t)}{dt} = g \left( a(t), p(t), t \right)
$$

where:
- $ a(t) $: Output of the network at time $ t $.
- $ p(t) $: Input to the network at time $ t $.
- $ g $: Nonlinear function defining the relationship between $ a(t) $, $ p(t) $, and $ t $.

# Stability Concepts in Recurrent Networks

To understand stability in neural networks, we start with some fundamental concepts using an intuitive example.

## Intuitive Example of Stability

Consider a **ball bearing** in a **gravitational field** with **dissipative friction**. We will examine three scenarios that illustrate different types of stability.

### 1. Asymptotically Stable Point

Imagine a **ball bearing in a trough**. 
- **Position $ a^* $**: The bottom of the trough, where the ball naturally rests.
- If we **move the ball to a different position**, it will **oscillate** but eventually settle back at $ a^* $ due to friction. 
- This point is called an **asymptotically stable point** because, even after displacement, the ball returns to the stable point.

### 2. Stable Point (Lyapunov Stability)

Now consider a **ball on a flat surface**.
- **Position at the center**: When positioned at the center, the ball remains stationary.
- If the ball is **moved away from the center**, it does not return, but it also **does not roll further away**.
- This point is **stable in the sense of Lyapunov**: small disturbances do not cause the ball to move away significantly.

### 3. Unstable Equilibrium Point

Finally, imagine the ball bearing at the **top of a hill**.
- **Equilibrium Position**: The ball remains at the top if undisturbed.
- However, **any slight disturbance** will cause it to roll down.
- This position is an **unstable equilibrium point**, as it cannot withstand even small perturbations.

## Stability Definitions and Positive Definite Functions

In this section, we will establish mathematical definitions of stability concepts in the context of equilibrium points. We focus on the **origin** $( a^* = 0 $) as the equilibrium point, where the derivative in Eq. (20.1) is zero. This assumption does not limit the generality of the discussion.

## Definition 1: Stability (in the sense of Lyapunov)

The **origin** is a stable equilibrium point if, for any value $ \epsilon > 0 $, there exists a number $ \delta(\epsilon) > 0 $ such that if:

$$
|a(0)| < \delta,
$$

then the resulting motion $ a(t) $ satisfies:

$$
|a(t)| < \epsilon \quad \text{for } t > 0.
$$

This means that the system output will not deviate too far from the stable point as long as it starts close to it.

- **Example**: A ball on a flat surface is stable in the sense of Lyapunov, as it will not roll away if displaced slightly.

## Definition 2: Asymptotic Stability

The **origin** is an asymptotically stable equilibrium point if there exists a $ \delta > 0 $ such that, whenever:

$$
|a(0)| < \delta,
$$

the resulting motion satisfies:

$$
a(t) \to 0 \quad \text{as } t \to \infty.
$$

This implies that, if the system starts within some distance $ \delta $ of the stable point, it will eventually converge to the stable point.

- **Example**: A ball in a trough with friction is asymptotically stable since it will eventually return to the bottom.

### Design Objective

For **Hopfield networks**, our objective is to create networks with many specified asymptotically stable points, each representing a unique prototype pattern.

## Definition 3: Positive Definite Function

A scalar function $ V(a) $ is **positive definite** if:

$$
V(0) = 0 \quad \text{and} \quad V(a) > 0 \quad \text{for } a \neq 0.
$$

## Definition 4: Positive Semidefinite Function

A scalar function $ V(a) $ is **positive semidefinite** if:

$$
V(a) \geq 0 \quad \text{for all } a.
$$

These definitions can be adapted to define **negative definite** and **negative semidefinite** functions.

# Lyapunov Stability Theorem

One of the most significant approaches to analyze the stability of nonlinear systems is through **Lyapunov's Stability Theorem**, developed by Alexandr Mikhailovich Lyapunov. Although Lyapunov's foundational work was published in 1892, it gained widespread recognition outside of Russia much later. Here, we focus on **Lyapunov's direct method** for stability analysis.

## Autonomous System

Consider the autonomous (unforced, no explicit time dependence) system:

$$
\frac{da}{dt} = g(a). \tag{20.2}
$$

## Theorem 1: Lyapunov Stability Theorem

The theorem can be stated as follows:

- **If a positive definite function** $ V(a) $ **can be found such that** $ \frac{dV(a)}{dt} $ **is negative semidefinite, then the origin** $ a = 0 $ **is stable** for the system described by Eq. (20.2).
- **If a positive definite function** $ V(a) $ **can be found such that** $ \frac{dV(a)}{dt} $ **is negative definite, then the origin** $ a = 0 $ **is asymptotically stable**.

In each case, $ V(a) $ is called a **Lyapunov function** of the system. This function can be thought of as a generalized energy function.

The idea is that if the system’s "energy" (as measured by $ V(a) $) continually decreases $( \frac{dV(a)}{dt} $ is negative definite), then it will eventually settle at a minimum energy state. Lyapunov's contribution was to generalize the notion of "energy," allowing the theorem to be applied even to systems where energy is hard to define or does not have a conventional meaning.

### Important Note

The theorem asserts that if we can find a suitable Lyapunov function \( V(a) \), the system is stable. However, if such a function cannot be found, this provides no information about the system's stability.

### Equation of Motion for the Pendulum

Using **Newton's Second Law** $( F = ma $), we write the equation of motion for the pendulum:

$$
ml^2 \frac{d^2 \theta}{dt^2} = -c \frac{d \theta}{dt} - mg \sin(\theta), \tag{20.3}
$$

where:
- $ ml^2 $ is the moment of inertia,
- $ c $ is the damping coefficient,
- $ g $ is the gravitational acceleration.

---

In the next section, we will explore how to apply Lyapunov’s theorem to analyze the stability of this pendulum system.



## Lyapunov Stability for the Pendulum

In this section, we will derive the Lyapunov function for the pendulum and analyze its stability using Lyapunov's theorem.

## Equilibrium Point

Starting with the equations of motion for the pendulum, we have:

$$
\frac{d^2 a_1}{dt^2} = -\frac{c}{ml} \frac{da_1}{dt} - \frac{g}{l} \sin(a_1) = 0. \tag{20.9}
$$

The derivatives are zero, indicating that the origin is an equilibrium point. 

## Lyapunov Function

We will use the total energy of the pendulum as the Lyapunov function \( V \). The total energy is the sum of the kinetic and potential energies, given by:

$$
V(a) = \frac{1}{2} ml \left( a_2^2 \right) + mgl \left( 1 - \cos(a_1) \right). \tag{20.10}
$$

## Time Derivative of the Lyapunov Function

To test the stability of the system, we need to evaluate the time derivative of \( V \):

$$
\frac{dV}{dt} = \frac{\partial V}{\partial a_1} \frac{da_1}{dt} + \frac{\partial V}{\partial a_2} \frac{da_2}{dt}. \tag{20.11}
$$

The partial derivatives of $ V(a) $ can be obtained from Eq. (20.10), and the derivatives of the two state variables are given by:

$$
\frac{da_1}{dt} = a_2, \quad \frac{da_2}{dt} = -\frac{g}{l} \sin(a_1) - \frac{c}{ml} a_2. \tag{20.6, 20.7}
$$

Substituting these into the equation, we have:

$$
\frac{dV}{dt} = \left(mgl \sin(a_1)\right) a_2 + \left(ml a_2\right) \left(-\frac{g}{l} \sin(a_1) - \frac{c}{ml} a_2\right). \tag{20.12}
$$

The $ \left(mgl \sin(a_1)\right) a_2 $ terms cancel, leading to:

$$
\frac{dV}{dt} = -c l \left( a_2^2 \right) \leq 0. \tag{20.13}
$$

## Stability Analysis

To show that the origin $ (a = 0) $ is asymptotically stable, we must demonstrate that this derivative is negative definite. The derivative is zero at the origin; however, it is also zero for any value of $ a_1 $ as long as $ a_2 = 0 $. Therefore, $ \frac{dV}{dt} $ is negative semidefinite rather than negative definite.

From Lyapunov's theorem, we conclude that the origin is a stable point. However, we cannot assert that it is asymptotically stable using this Lyapunov function alone. We will explore a refinement of Lyapunov's theorem, **LaSalle's Invariance Theorem**, in the next section.

## Initial Conditions

The behavior of the pendulum shown in the graphs depends on the initial conditions of the two state variables. A different set of initial conditions may yield entirely different results, which we will explore further in the next section.



```python
import matplotlib.pyplot as plt
import math

# Pendulum parameters
g = 9.8    # acceleration due to gravity (m/s^2)
m = 1.0    # mass of the pendulum (kg)
l = 9.8    # length of the pendulum (m)
c = 1.96   # damping coefficient

# Time parameters
t_max = 40.0  # total time (s)
dt = 0.01     # time step (s)
time_steps = int(t_max / dt)

# Initialize state variables: [theta, theta_dot]
a1 = [0] * time_steps  # angle (radians)
a2 = [0] * time_steps  # angular velocity (radians/s)

# Initial conditions
a1[0] = 1.3  # initial angle
a2[0] = 1.3  # initial angular velocity

# Time integration using Euler's method
for t in range(1, time_steps):
    a1[t] = a1[t-1] + a2[t-1] * dt
    a2[t] = a2[t-1] - (g/l) * math.sin(a1[t-1]) * dt - c * a2[t-1] * dt

# Calculate the Lyapunov function (energy)
V = [0] * time_steps
for t in range(time_steps):
    V[t] = 0.5 * m * (l * a2[t])**2 + m * g * l * (1 - math.cos(a1[t]))

# Plotting the state variables
plt.figure(figsize=(12, 8))

# State variables plot
plt.subplot(3, 1, 1)
plt.plot([t * dt for t in range(time_steps)], a1, label=r'$\theta (t)$', color='blue')
plt.plot([t * dt for t in range(time_steps)], a2, label=r'$\dot{\theta} (t)$', color='red')
plt.title('State Variables vs Time')
plt.xlabel('Time (s)')
plt.ylabel('State Variables')
plt.legend()
plt.grid()

# Energy plot
plt.subplot(3, 1, 2)
plt.plot([t * dt for t in range(time_steps)], V, color='green')
plt.title('Pendulum Energy vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Energy (Joules)')
plt.grid()

# Contour plot
plt.subplot(3, 1, 3)
plt.contourf(a1, a2, V, levels=20, cmap='viridis')
plt.colorbar(label='Energy (Joules)')
plt.title('Pendulum Energy Contours')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\dot{\theta}$ (radians/s)')

plt.tight_layout()
plt.show()
```

## LaSalle's Invariance Theorem

The pendulum example demonstrated a problem with Lyapunov's theorem. We found a Lyapunov function whose derivative was only negative semidefinite (not negative definite), and yet we know that the origin is asymptotically stable for the pendulum system. In this section, we will introduce a theorem that clarifies this uncertainty in Lyapunov's theorem. It does so by defining those regions of the state space where the derivative of the Lyapunov function is zero and then identifying those parts of that region that can trap the trajectory.

## Definitions

**Definition 5: Lyapunov Function**

Let $ V $ be a continuously differentiable function from $ \mathbb{R}^n $ to $ \mathbb{R} $. If $ G $ is any subset of $ \mathbb{R}^n $, we say that $ V $ is a Lyapunov function on $ G $ for the system 

$$
\frac{da}{dt} = g(a)
$$

if 

$$
\frac{dV(a)}{dt} = \nabla V(a) \cdot g(a) \tag{20.18}
$$

does not change sign on $ G $. This is a generalization of our previous definition of the Lyapunov function, which we used in Theorem 1. Here we do not require that the function be positive definite. In fact, there is no direct requirement on the function itself (except that it be continuously differentiable). The only requirement is on the derivative of $ V $. The derivative cannot change sign anywhere on the set $ G $. Note that the derivative will not change sign if it is negative semidefinite or if it is positive semidefinite.

**Definition 6: Set \( Z \)**

$$
Z = \{ a : \frac{dV(a)}{dt} = 0, \, a \in \text{closure of } G \} \tag{20.19}
$$

Here "the closure of \( G \)" includes the interior and the boundary of \( G \). This is a key set. It contains all of those points where the derivative of the Lyapunov function is zero. Later we will want to determine where in this set the system trajectory can be trapped.

**Definition 7: Invariant Set**

A set of points in $ \mathbb{R}^n $ is invariant with respect to 

$$
\frac{da}{dt} = g(a)
$$

if every solution of 

$$
\frac{da}{dt} = g(a)
$$

starting in that set remains in the set for all time. If the system gets into an invariant set, then it can't get out.

**Definition 8: Set $ L $**

$ L $ is defined as the largest invariant set in $ Z $. This set includes all possible points at which the solution might converge. The Lyapunov function does not change in $ L $ (because its derivative is zero), and the trajectory will be trapped in $ L $ (because it is an invariant set). Now, if this set has only one stable point, then that point is asymptotically stable.

## Theorem

**LaSalle's Invariance Theorem** extends the Lyapunov Stability Theorem. We will use it to design Hopfield networks in the next chapter. The theorem proceeds as follows [Lasa67].

**Theorem 2: LaSalle's Invariance Theorem**

If $ V $ is a Lyapunov function on $ G $ for 

$$
\frac{da}{dt} = g(a)
$$

then each solution $ a(t) $ that remains in $ G $ for all $ t > 0 $ approaches 

$$
L^0 = L \cup \{ \infty \} \quad \text{as } t \to \infty. 
$$

$ G $ is a basin of attraction for $ L $, which has all of the stable points. If all trajectories are bounded, then $ a(t) \to L $ as $ t \to \infty $. If a trajectory stays in $ G $, then it will either converge to $ L $ or it will go to infinity. If all trajectories are bounded, then all trajectories will converge to $ L $.

There is a corollary to LaSalle's theorem that we will use extensively. It involves choosing the set $ G $ in a special way.


# LaSalle's Corollary

**Corollary 1: LaSalle's Corollary**  
Let $ G $ be a component (one connected subset) of $ \Omega_\eta = \{ a : V(a) < \eta \} $. Assume that $ G $ is bounded, $ \frac{dV(a)}{dt} \leq 0 $ on the set $ G $, and let the set $$ L^\circ = \text{closure}(L \cap G) $$ be a subset of \( G \). Then $ L^\circ $ is an attractor, and $ G $ is in its region of attraction.

LaSalle's theorem, and its corollary, are very powerful. Not only can they tell us which points are stable ($ L^\circ $), but they can also provide us with a partial region of attraction ($ G $). (Note that $ L^\circ $ is constructed differently in the corollary than in the theorem.)

To clarify LaSalle's Invariance Theorem, let’s return to the pendulum example.

## Example: Pendulum

Let’s apply Corollary 1 to the pendulum example. The first step in using the corollary will be to choose the set $ \Omega_\eta $. This set will then be used to select the set $ G $ (a component of $ \Omega_\eta $). For this example, we will use the value $ \eta = 100 $, therefore 

$$
\Omega_{100} = \{ a : V(a) \leq 100 \}
$$

This set is displayed in blue in Figure 20.8.

$\Omega_{100}$

Next, we need to choose a component (connected subset) of $ \Omega_{100} $ for the set $ G $. Since we have been investigating the stability of the origin, let’s choose the component of $ \Omega_{100} $ that contains $ a = 0 $. The resulting set is shown in Fig.9.



Fig.8 Illustration of the Set $\Omega$ 100

Now that we have chosen \( G \), we need to check that the derivative of the Lyapunov function is less than or equal to zero on \( G \). From Eq. (20.17), we know that 

$$
\frac{dV(a)}{dt}
$$

is negative semidefinite. Therefore, it will certainly be less than or equal to zero on \( G \).

We are now ready to determine the attractor set $ L^\circ $. We begin with the set $ L $, which is the largest invariant set in 

$$
Z = \{ a : \frac{dV(a)}{dt} = 0, a \text{ in the closure of } G \} = \{ a : a_2 = 0, a \text{ in the closure of } G \}
$$

This can also be written as 

$$
Z = \{ a : a_2 = 0, -1.6 \leq a_1 \leq 1.6 \}.
$$

We know from Eq. (20.17) that the derivative of $ V(a) $ is only zero when the velocity is zero, which corresponds to the $ a_1 $ axis. Therefore, $ Z $ consists of the segment of the $ a_1 $ axis that falls within $ G $. The set $ Z $ is displayed in Fig.10.



Fig.9 Illustration of the Set G


The set $ L $ is the largest invariant set in $ Z $. To find $ L $, we need to answer the question: If we start the pendulum from an initial position between $ -1.6 $ and $ 1.6 $ radians, with zero initial velocity, will the velocity of the pendulum remain zero? 

Clearly, the only such initial condition would be $ 0 $ radians (straight down). If we start the pendulum from any other position in $ Z $, the pendulum will start to fall, so the velocity will not remain zero, and the trajectory will move out of $ Z $. Therefore, the set $ L $ consists only of the origin:

$$
L = \{ a : a = 0 \}.
$$








Fig.10 Illustration of the Set Z


## Illustration of the Set $ Z $

The set $L^\circ $ is the closure of the intersection of $ L $ and $ G $, which in this case is simply $ L $:

$$
L^\circ = \text{closure}(L \cap G) = L = \{ a : a = 0 \}. \tag{20.25}
$$

Therefore, based on LaSalle's corollary, $ L^\circ $ is an attractor (asymptotically stable point), and $ G $ is in its region of attraction. This means that any trajectory that starts in $ G $ will decay to the origin.

Now suppose that we had taken a bigger region for $ \Omega_\eta $, such as 

$$
\Omega_{300} = \{ a : V(a) < 300 \}. \tag{20.26}
$$

This set is shown in gray in Fig.11.




Fig.11 Illustration of $G =  \Omega_300 $ (Gray) and Z

We let $ G = \Omega_{300} $, since $ \Omega_{300} $ has only one component. The set $ Z $ is given by 

$$
Z = \{ a : a_2 = 0 \}, \tag{20.27}
$$ 

which is shown by the blue bar on the horizontal axis of Fig.11. Thus, it follows that 

$$
L^\circ = L = \{ a : a_1 = \pm n\pi, a_2 = 0 \}. \tag{20.28}
$$

This is because there are now several different positions within the set $ Z $ where we can place the pendulum without causing the velocity to become nonzero. The pendulum can be pointing directly up or directly down. This corresponds to the positions $ \pm n\pi $ for any integer $ n $. If we place the pendulum in any of these positions, with zero velocity, then the pendulum will remain stationary. We can show this by setting the derivatives equal to zero in Eq. (20.14) and Eq. (20.15):

$$
\frac{da_1}{dt} = a_2 = 0, \tag{20.29}
$$

$$
\frac{da_2}{dt} = -\sin(a) - 0.2a = -\sin(a) = 0 \Rightarrow (a = \pm n\pi). \tag{20.30}
$$

For this choice of $ G = \Omega_{300} $, we can say very little about where the trajectory will converge. We tried to increase the size of our known region of attraction for the origin, but this $ G $ is a region of attraction for all of the equilibrium points. We made $ G $ too large. The set $ L^\circ $ is illustrated by the blue dots in Fig.12.


Fig.12 The Set $^*$

We cannot tell which of the equilibrium points (blue dots) will attract the trajectory. All we can say is that if we start somewhere in $ \Omega_{300} $, one of the equilibrium points will attract the system solution, but we cannot say for sure which one it will be.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set(style="whitegrid")

# Define pendulum parameters
g = 9.81  # acceleration due to gravity (m/s^2)
L = 1.0   # length of pendulum (m)
b = 0.2   # damping coefficient

# Simple sine approximation for small angles
def simple_sin(x):
    return x  # For small angles (in radians), sin(x) ≈ x

# Function to compute the derivative of the system
def pendulum_deriv(a1, a2, dt):
    # Calculate angular acceleration using the sine approximation
    a2_new = a2 - (b * a2 + (g / L) * simple_sin(a1)) * dt
    a1_new = a1 + a2_new * dt
    return a1_new, a2_new

# Simulate the pendulum
def simulate_pendulum(initial_conditions, total_time, dt):
    time_steps = int(total_time / dt)
    a1_values = []
    a2_values = []
    
    a1, a2 = initial_conditions
    for _ in range(time_steps):
        a1_values.append(a1)
        a2_values.append(a2)
        a1, a2 = pendulum_deriv(a1, a2, dt)
    
    return a1_values, a2_values

# Define time parameters
dt = 0.05  # time step (s)
total_time = 10  # total simulation time (s)

# Initial conditions: [angle (rad), angular velocity (rad/s)]
initial_conditions = [2.0, 0.0]  # Start at 2 radians with 0 velocity

# Run the simulation
a1_values, a2_values = simulate_pendulum(initial_conditions, total_time, dt)

# Calculate the Lyapunov function values
def lyapunov_function(a1, a2):
    return 0.5 * (a1**2 + a2**2)  # V(a) = 1/2 * (a1^2 + a2^2)

# Calculate Lyapunov function over time
lyapunov_values = [lyapunov_function(a1, a2) for a1, a2 in zip(a1_values, a2_values)]

# Create a plot of the trajectory in phase space
plt.figure(figsize=(12, 6))
sns.lineplot(x=a1_values, y=a2_values, label='Trajectory in Phase Space', color='blue')
plt.title('Pendulum Trajectory in Phase Space')
plt.xlabel('Angle (a1) [radians]')
plt.ylabel('Angular Velocity (a2) [radians/s]')
plt.axhline(0, color='grey', lw=1, ls='--')  # Horizontal line at a2 = 0
plt.axvline(0, color='grey', lw=1, ls='--')  # Vertical line at a1 = 0
plt.grid()
plt.legend()
plt.show()

# Create a plot of the Lyapunov function over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=[i * dt for i in range(len(lyapunov_values))], y=lyapunov_values, label='Lyapunov Function (V)', color='orange')
plt.title('Lyapunov Function over Time')
plt.xlabel('Time [s]')
plt.ylabel('Lyapunov Function Value (V)')
plt.axhline(0, color='grey', lw=1, ls='--')  # Horizontal line at V = 0
plt.grid()
plt.legend()
plt.show()
```
