## Particle Swarm Optimization (PSO)

Particle Swarm Optimization (PSO) is a global optimization technique introduced by Eberhart and Kennedy in 1995. PSO is a population-based stochastic algorithm for finding optimal regions in complex search spaces through the interaction of individuals (called particles) in the population. Each particle represents a potential solution to an optimization problem, and the particles adjust their positions based on their own experience and the experience of their neighbors.

### Basic Concepts

PSO is based on two main disciplines: **social science** and **computer science**, particularly the principles of **swarm intelligence**. Swarm intelligence is a concept inspired by the collective behavior of social animals such as birds, bees, and fish, where each agent (or particle) interacts with others to improve the global outcome.

### Swarm Intelligence Principles:
- **Proximity principle**: Particles can carry out simple space-time computations.
- **Quality principle**: Particles respond to quality factors in the environment.
- **Diverse response principle**: Particles avoid overly narrow channels of behavior.
- **Stability principle**: Particles maintain stable behaviors even when the environment changes.
- **Adaptability principle**: Particles adjust their behavior when necessary.

### Computational Characteristics:
PSO, like cellular automata, updates individual particles (cells) in parallel, and each particle's new value depends on its previous value and its neighbors' values.

## Canonical Particle Swarm

Given a data set $ Z = \{ z_1, z_2, \dots, z_N \} $ representing $ N $ patterns $ z_i \in \mathbb{R}^D $, the objective of PSO is to partition the data into $ K $ clusters, such that:

$$
C_k \neq \emptyset, \quad \forall k = 1, 2, \dots, K
$$
$$
C_k \cap C_l = \emptyset, \quad \forall k, l = 1, 2, \dots, K \quad (k \neq l)
$$
$$
\bigcup_{k=1}^K C_k = Z
$$

This clustering task can be framed as an optimization problem where the objective is to minimize the fitness function $ f(C_k) $ for each cluster $ C_k $, $ \forall k = 1, 2, \dots, K $.

### PSO Parameters:
- $ N_p $: The swarm (or population) size.
- $ x_i(t) $: The position vector of the $ i $-th particle at time $ t $.
- $ v_i(t) $: The velocity vector of the $ i $-th particle at time $t $.
- $ p_{i,best}(t) $: The best position of the $ i $-th particle up to time $ t $.
- $ g_{best} $: The global best position found by any particle during all previous steps.
- $ \alpha $: The inertia weight or constriction coefficient.
- $ \beta $: The acceleration constant.
- $ U(0, \beta) $: A uniform random number generator.

### PSO Update Rules:

1. **Velocity Update**:
   $$
   v_i(t+1) = \alpha v_i(t) + U(0, \beta) \left( p_{i,best}(t) - x_i(t) \right) + U(0, \beta) \left( g_{best} - x_i(t) \right)
   $$

2. **Position Update**:
   $$
   x_i(t+1) = x_i(t) + v_i(t+1)
   $$

3. **Personal Best Update**:
   $$
   p_{i,best}(t+1) = 
   \begin{cases}
   x_i(t+1), & \text{if } f(x_i(t+1)) < f(p_{i,best}(t)) \\
   p_{i,best}(t), & \text{otherwise}
   \end{cases}
   $$

4. **Global Best Update**:
   $$
   g_{best} = \arg \min_{i} f(p_{i,best}(t+1))
   $$

### Algorithm Steps:

The basic steps of the PSO algorithm are as follows:

1. **Initialization**:
   - Initialize the positions $ x_1(1), \dots, x_{N_p}(1) $ and velocities $ v_1(1), \dots, v_{N_p}(1) $ of the particles randomly.
   - Set the personal best positions $ p_{i,best}(1) = x_i(1) $ for each particle.
   - Set the global best position $ g_{best} $ to the particle with the best fitness.

2. **Repeat for each iteration**:
   - For each particle:
     - Update the velocity using the equation above.
     - Update the position using the updated velocity.
     - Update the personal best if the new position has a better fitness.
   - Update the global best position.

3. **Termination**:
   - Stop when a stopping criterion is met (e.g., the maximum number of iterations or if $ g_{best} $ is sufficiently good).

### PSO Algorithm:
$$
\text{1. Input: cost function } f, \text{ number of particles } N_p, \text{ maximum cycles } N_c, \text{ dimension } D, \text{ inertia weight } \alpha, \text{ acceleration constant } \beta.
$$

### Particle Swarm Optimization (PSO) Algorithm

The PSO algorithm can be described as follows:

1. **Input**: 
    - Cost function $ f $
    - Number of particles $ N_p $
    - Maximum cycles $ N_c $
    - The dimension $ D $ of position and velocity vectors
    - The inertia weight $ \alpha $
    - Acceleration constants $ \beta_1 $, $ \beta_2 

2. **Initialization**:
    - Create particle positions $ x_1(1), x_2(1), \ldots, x_{N_p}(1) $ and particle velocities $ v_1(1), v_2(1), \ldots, v_{N_p}(1) $
    - Set $ p_{i,\text{best}}(1) = x_i(1) $, for $ i = 1, 2, \ldots, N_p $
    - Set $ g_{\text{best}} = x_i(1) $, where $ i = \arg \min \{ f(x_1(1)), f(x_2(1)), \ldots, f(x_{N_p}(1)) \} $

3. **Repeat**:
    - For $ t = 1, 2, \ldots, N_c $ do:
        1. For $ i = 1, 2, \ldots, N_p $ do:
            - Update velocity:
            $$
            v_i(t+1) = \alpha v_i(t) + U(0, \beta_1) \left( p_{i,\text{best}}(t) - x_i(t) \right) + U(0, \beta_2) \left( g_{\text{best}} - x_i(t) \right)
            $$
            - Update position:
            $$
            x_i(t+1) = x_i(t) + v_i(t+1)
            $$
            - Update personal best:
            $$
            p_{i,\text{best}}(t+1) = 
            \begin{cases} 
            x_i(t+1) & \text{if } f(x_i(t+1)) < f(p_{i,\text{best}}(t)) \\
            p_{i,\text{best}}(t) & \text{otherwise}
            \end{cases}
            $$
        2. Update global best:
       $$
        g_{\text{best}} = \arg \min \{ f(p_{j,\text{best}}(t+1)) \}, \quad j = 1, 2, \ldots, N_p
        $$
    - Until $ g_{\text{best}} $ is sufficiently good or the maximum cycle is reached.

4. **Output**: Global best position $ g_{\text{best}} $



```python
import numpy as np

# Define the cost function (example: Sphere function)
def cost_function(x):
    return np.sum(x**2)

# Particle Swarm Optimization (PSO) implementation
class PSO:
    def __init__(self, cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound):
        self.cost_function = cost_function
        self.Np = Np  # Number of particles
        self.Nc = Nc  # Maximum cycles
        self.D = D  # Dimension of position and velocity
        self.alpha = alpha  # Inertia weight
        self.beta = beta  # Acceleration constant
        self.lower_bound = lower_bound  # Lower bound for particles' positions
        self.upper_bound = upper_bound  # Upper bound for particles' positions

        # Initialize particle positions and velocities
        self.positions = np.random.uniform(lower_bound, upper_bound, (Np, D))
        self.velocities = np.random.uniform(-1, 1, (Np, D))
        
        # Initialize personal best positions and global best position
        self.p_best = np.copy(self.positions)
        self.p_best_values = np.apply_along_axis(self.cost_function, 1, self.p_best)
        self.g_best = self.p_best[np.argmin(self.p_best_values)]
        
    def update_velocity(self, i, t):
        # Update particle velocity
        inertia = self.alpha * self.velocities[i]
        cognitive = self.beta * np.random.random() * (self.p_best[i] - self.positions[i])
        social = self.beta * np.random.random() * (self.g_best - self.positions[i])
        self.velocities[i] = inertia + cognitive + social

    def update_position(self, i):
        # Update particle position
        self.positions[i] += self.velocities[i]
        # Ensure positions stay within bounds
        self.positions[i] = np.clip(self.positions[i], self.lower_bound, self.upper_bound)
        
    def update_personal_best(self, i):
        # Update personal best if the current position is better
        current_value = self.cost_function(self.positions[i])
        if current_value < self.p_best_values[i]:
            self.p_best[i] = np.copy(self.positions[i])
            self.p_best_values[i] = current_value

    def update_global_best(self):
        # Update global best position
        best_particle_idx = np.argmin(self.p_best_values)
        self.g_best = np.copy(self.p_best[best_particle_idx])
        
    def optimize(self):
        for t in range(self.Nc):
            for i in range(self.Np):
                # Update the velocity, position, and personal best of each particle
                self.update_velocity(i, t)
                self.update_position(i)
                self.update_personal_best(i)
            # Update global best
            self.update_global_best()

            # Print the global best cost and position for each iteration
            print(f"Cycle {t+1}: Global Best Cost = {self.cost_function(self.g_best)}, Global Best Position = {self.g_best}")
            
        return self.g_best

# Parameters
Np = 30  # Number of particles
Nc = 100  # Maximum number of cycles
D = 5  # Dimension of position and velocity
alpha = 0.9  # Inertia weight
beta = 1.5  # Acceleration constant
lower_bound = -5  # Lower bound for particle positions
upper_bound = 5  # Upper bound for particle positions

# Initialize and run the PSO algorithm
pso = PSO(cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound)
best_position = pso.optimize()
print(f"Final Global Best Position: {best_position}")
```

```python
import random
import matplotlib.pyplot as plt

# Define the cost function (example: Sphere function)
def cost_function(x):
    return sum([xi**2 for xi in x])

# Particle Swarm Optimization (PSO) implementation
class PSO:
    def __init__(self, cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound):
        self.cost_function = cost_function
        self.Np = Np  # Number of particles
        self.Nc = Nc  # Maximum cycles
        self.D = D  # Dimension of position and velocity
        self.alpha = alpha  # Inertia weight
        self.beta = beta  # Acceleration constant
        self.lower_bound = lower_bound  # Lower bound for particles' positions
        self.upper_bound = upper_bound  # Upper bound for particles' positions

        # Initialize particle positions and velocities
        self.positions = [[random.uniform(lower_bound, upper_bound) for _ in range(D)] for _ in range(Np)]
        self.velocities = [[random.uniform(-1, 1) for _ in range(D)] for _ in range(Np)]
        
        # Initialize personal best positions and global best position
        self.p_best = [list(position) for position in self.positions]
        self.p_best_values = [self.cost_function(position) for position in self.p_best]
        self.g_best = self.p_best[self.p_best_values.index(min(self.p_best_values))]
        
        # List to store the global best cost over cycles
        self.global_best_costs = []

    def update_velocity(self, i, t):
        # Update particle velocity
        inertia = [self.alpha * v for v in self.velocities[i]]
        cognitive = [self.beta * random.random() * (p - x) for p, x in zip(self.p_best[i], self.positions[i])]
        social = [self.beta * random.random() * (g - x) for g, x in zip(self.g_best, self.positions[i])]
        self.velocities[i] = [inertia[j] + cognitive[j] + social[j] for j in range(self.D)]

    def update_position(self, i):
        # Update particle position
        self.positions[i] = [self.positions[i][j] + self.velocities[i][j] for j in range(self.D)]
        # Ensure positions stay within bounds
        self.positions[i] = [max(self.lower_bound, min(self.upper_bound, x)) for x in self.positions[i]]
        
    def update_personal_best(self, i):
        # Update personal best if the current position is better
        current_value = self.cost_function(self.positions[i])
        if current_value < self.p_best_values[i]:
            self.p_best[i] = list(self.positions[i])
            self.p_best_values[i] = current_value

    def update_global_best(self):
        # Update global best position
        best_particle_idx = self.p_best_values.index(min(self.p_best_values))
        self.g_best = list(self.p_best[best_particle_idx])
        
    def optimize(self):
        for t in range(self.Nc):
            for i in range(self.Np):
                # Update the velocity, position, and personal best of each particle
                self.update_velocity(i, t)
                self.update_position(i)
                self.update_personal_best(i)
            # Update global best
            self.update_global_best()

            # Store the global best cost for plotting
            self.global_best_costs.append(self.cost_function(self.g_best))
            
            # Print the global best cost and position for each iteration
            print(f"Cycle {t+1}: Global Best Cost = {self.cost_function(self.g_best)}, Global Best Position = {self.g_best}")
            
        return self.g_best

# Parameters
Np = 30  # Number of particles
Nc = 100  # Maximum number of cycles
D = 5  # Dimension of position and velocity
alpha = 0.9  # Inertia weight
beta = 1.5  # Acceleration constant
lower_bound = -5  # Lower bound for particle positions
upper_bound = 5  # Upper bound for particle positions

# Initialize and run the PSO algorithm
pso = PSO(cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound)
best_position = pso.optimize()

# Plot the global best cost over cycles
plt.plot(range(Nc), pso.global_best_costs)
plt.xlabel('Cycle')
plt.ylabel('Global Best Cost')
plt.title('PSO Optimization Progress')
plt.show()

# Print the final global best position
print(f"Final Global Best Position: {best_position}")
```
