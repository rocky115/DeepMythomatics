## Particle Swarm Optimization (PSO)

Particle Swarm Optimization (PSO) is a global optimization technique introduced by Eberhart and Kennedy in 1995. PSO is a population-based stochastic algorithm for finding optimal regions in complex search spaces through the interaction of individuals (called particles) in the population. Each particle represents a potential solution to an optimization problem, and the particles adjust their positions based on their own experience and the experience of their neighbors.

### Basic Concepts

PSO is based on two main disciplines: **social science** and **computer science**, particularly the principles of **swarm intelligence**. Swarm intelligence is a concept inspired by the collective behavior of social animals such as birds, bees, and fish, where each agent (or particle) interacts with others to improve the global outcome.

### Swarm Intelligence Principles:
- **Proximity principle**: Particles can carry out simple space-time computations.
- **Quality principle**: Particles respond to quality factors in the environment.
- **Diverse response principle**: Particles avoid overly narrow channels of behavior.
- **Stability principle**: Particles maintain stable behaviors even when the environment changes.
- **Adaptability principle**: Particles adjust their behavior when necessary.

### Computational Characteristics:
PSO, like cellular automata, updates individual particles (cells) in parallel, and each particle's new value depends on its previous value and its neighbors' values.

## Canonical Particle Swarm

Given a data set $ Z = \{ z_1, z_2, \dots, z_N \} $ representing $ N $ patterns $ z_i \in \mathbb{R}^D $, the objective of PSO is to partition the data into $ K $ clusters, such that:

$$
C_k \neq \emptyset, \quad \forall k = 1, 2, \dots, K
$$
$$
C_k \cap C_l = \emptyset, \quad \forall k, l = 1, 2, \dots, K \quad (k \neq l)
$$
$$
\bigcup_{k=1}^K C_k = Z
$$

This clustering task can be framed as an optimization problem where the objective is to minimize the fitness function $ f(C_k) $ for each cluster $ C_k $, $ \forall k = 1, 2, \dots, K $.

### PSO Parameters:
- $ N_p $: The swarm (or population) size.
- $ x_i(t) $: The position vector of the $ i $-th particle at time $ t $.
- $ v_i(t) $: The velocity vector of the $ i $-th particle at time $t $.
- $ p_{i,best}(t) $: The best position of the $ i $-th particle up to time $ t $.
- $ g_{best} $: The global best position found by any particle during all previous steps.
- $ \alpha $: The inertia weight or constriction coefficient.
- $ \beta $: The acceleration constant.
- $ U(0, \beta) $: A uniform random number generator.

### PSO Update Rules:

1. **Velocity Update**:
   $$
   v_i(t+1) = \alpha v_i(t) + U(0, \beta) \left( p_{i,best}(t) - x_i(t) \right) + U(0, \beta) \left( g_{best} - x_i(t) \right)
   $$

2. **Position Update**:
   $$
   x_i(t+1) = x_i(t) + v_i(t+1)
   $$

3. **Personal Best Update**:
   $$
   p_{i,best}(t+1) = 
   \begin{cases}
   x_i(t+1), & \text{if } f(x_i(t+1)) < f(p_{i,best}(t)) \\
   p_{i,best}(t), & \text{otherwise}
   \end{cases}
   $$

4. **Global Best Update**:
   $$
   g_{best} = \arg \min_{i} f(p_{i,best}(t+1))
   $$

### Algorithm Steps:

The basic steps of the PSO algorithm are as follows:

1. **Initialization**:
   - Initialize the positions $ x_1(1), \dots, x_{N_p}(1) $ and velocities $ v_1(1), \dots, v_{N_p}(1) $ of the particles randomly.
   - Set the personal best positions $ p_{i,best}(1) = x_i(1) $ for each particle.
   - Set the global best position $ g_{best} $ to the particle with the best fitness.

2. **Repeat for each iteration**:
   - For each particle:
     - Update the velocity using the equation above.
     - Update the position using the updated velocity.
     - Update the personal best if the new position has a better fitness.
   - Update the global best position.

3. **Termination**:
   - Stop when a stopping criterion is met (e.g., the maximum number of iterations or if $ g_{best} $ is sufficiently good).

### PSO Algorithm:
$$
\text{1. Input: cost function } f, \text{ number of particles } N_p, \text{ maximum cycles } N_c, \text{ dimension } D, \text{ inertia weight } \alpha, \text{ acceleration constant } \beta.
$$

### Particle Swarm Optimization (PSO) Algorithm

The PSO algorithm can be described as follows:

1. **Input**: 
    - Cost function $ f $
    - Number of particles $ N_p $
    - Maximum cycles $ N_c $
    - The dimension $ D $ of position and velocity vectors
    - The inertia weight $ \alpha $
    - Acceleration constants $ \beta_1 $, $ \beta_2 

2. **Initialization**:
    - Create particle positions $ x_1(1), x_2(1), \ldots, x_{N_p}(1) $ and particle velocities $ v_1(1), v_2(1), \ldots, v_{N_p}(1) $
    - Set $ p_{i,\text{best}}(1) = x_i(1) $, for $ i = 1, 2, \ldots, N_p $
    - Set $ g_{\text{best}} = x_i(1) $, where $ i = \arg \min \{ f(x_1(1)), f(x_2(1)), \ldots, f(x_{N_p}(1)) \} $

3. **Repeat**:
    - For $ t = 1, 2, \ldots, N_c $ do:
        1. For $ i = 1, 2, \ldots, N_p $ do:
            - Update velocity:
            $$
            v_i(t+1) = \alpha v_i(t) + U(0, \beta_1) \left( p_{i,\text{best}}(t) - x_i(t) \right) + U(0, \beta_2) \left( g_{\text{best}} - x_i(t) \right)
            $$
            - Update position:
            $$
            x_i(t+1) = x_i(t) + v_i(t+1)
            $$
            - Update personal best:
            $$
            p_{i,\text{best}}(t+1) = 
            \begin{cases} 
            x_i(t+1) & \text{if } f(x_i(t+1)) < f(p_{i,\text{best}}(t)) \\
            p_{i,\text{best}}(t) & \text{otherwise}
            \end{cases}
            $$
        2. Update global best:
       $$
        g_{\text{best}} = \arg \min \{ f(p_{j,\text{best}}(t+1)) \}, \quad j = 1, 2, \ldots, N_p
        $$
    - Until $ g_{\text{best}} $ is sufficiently good or the maximum cycle is reached.

4. **Output**: Global best position $ g_{\text{best}} $



````python
import numpy as np

# Define the cost function (example: Sphere function)
def cost_function(x):
    return np.sum(x**2)

# Particle Swarm Optimization (PSO) implementation
class PSO:
    def __init__(self, cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound):
        self.cost_function = cost_function
        self.Np = Np  # Number of particles
        self.Nc = Nc  # Maximum cycles
        self.D = D  # Dimension of position and velocity
        self.alpha = alpha  # Inertia weight
        self.beta = beta  # Acceleration constant
        self.lower_bound = lower_bound  # Lower bound for particles' positions
        self.upper_bound = upper_bound  # Upper bound for particles' positions

        # Initialize particle positions and velocities
        self.positions = np.random.uniform(lower_bound, upper_bound, (Np, D))
        self.velocities = np.random.uniform(-1, 1, (Np, D))
        
        # Initialize personal best positions and global best position
        self.p_best = np.copy(self.positions)
        self.p_best_values = np.apply_along_axis(self.cost_function, 1, self.p_best)
        self.g_best = self.p_best[np.argmin(self.p_best_values)]
        
    def update_velocity(self, i, t):
        # Update particle velocity
        inertia = self.alpha * self.velocities[i]
        cognitive = self.beta * np.random.random() * (self.p_best[i] - self.positions[i])
        social = self.beta * np.random.random() * (self.g_best - self.positions[i])
        self.velocities[i] = inertia + cognitive + social

    def update_position(self, i):
        # Update particle position
        self.positions[i] += self.velocities[i]
        # Ensure positions stay within bounds
        self.positions[i] = np.clip(self.positions[i], self.lower_bound, self.upper_bound)
        
    def update_personal_best(self, i):
        # Update personal best if the current position is better
        current_value = self.cost_function(self.positions[i])
        if current_value < self.p_best_values[i]:
            self.p_best[i] = np.copy(self.positions[i])
            self.p_best_values[i] = current_value

    def update_global_best(self):
        # Update global best position
        best_particle_idx = np.argmin(self.p_best_values)
        self.g_best = np.copy(self.p_best[best_particle_idx])
        
    def optimize(self):
        for t in range(self.Nc):
            for i in range(self.Np):
                # Update the velocity, position, and personal best of each particle
                self.update_velocity(i, t)
                self.update_position(i)
                self.update_personal_best(i)
            # Update global best
            self.update_global_best()

            # Print the global best cost and position for each iteration
            print(f"Cycle {t+1}: Global Best Cost = {self.cost_function(self.g_best)}, Global Best Position = {self.g_best}")
            
        return self.g_best

# Parameters
Np = 30  # Number of particles
Nc = 100  # Maximum number of cycles
D = 5  # Dimension of position and velocity
alpha = 0.9  # Inertia weight
beta = 1.5  # Acceleration constant
lower_bound = -5  # Lower bound for particle positions
upper_bound = 5  # Upper bound for particle positions

# Initialize and run the PSO algorithm
pso = PSO(cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound)
best_position = pso.optimize()
print(f"Final Global Best Position: {best_position}")
````

````python
import random
import matplotlib.pyplot as plt

# Define the cost function (example: Sphere function)
def cost_function(x):
    return sum([xi**2 for xi in x])

# Particle Swarm Optimization (PSO) implementation
class PSO:
    def __init__(self, cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound):
        self.cost_function = cost_function
        self.Np = Np  # Number of particles
        self.Nc = Nc  # Maximum cycles
        self.D = D  # Dimension of position and velocity
        self.alpha = alpha  # Inertia weight
        self.beta = beta  # Acceleration constant
        self.lower_bound = lower_bound  # Lower bound for particles' positions
        self.upper_bound = upper_bound  # Upper bound for particles' positions

        # Initialize particle positions and velocities
        self.positions = [[random.uniform(lower_bound, upper_bound) for _ in range(D)] for _ in range(Np)]
        self.velocities = [[random.uniform(-1, 1) for _ in range(D)] for _ in range(Np)]
        
        # Initialize personal best positions and global best position
        self.p_best = [list(position) for position in self.positions]
        self.p_best_values = [self.cost_function(position) for position in self.p_best]
        self.g_best = self.p_best[self.p_best_values.index(min(self.p_best_values))]
        
        # List to store the global best cost over cycles
        self.global_best_costs = []

    def update_velocity(self, i, t):
        # Update particle velocity
        inertia = [self.alpha * v for v in self.velocities[i]]
        cognitive = [self.beta * random.random() * (p - x) for p, x in zip(self.p_best[i], self.positions[i])]
        social = [self.beta * random.random() * (g - x) for g, x in zip(self.g_best, self.positions[i])]
        self.velocities[i] = [inertia[j] + cognitive[j] + social[j] for j in range(self.D)]

    def update_position(self, i):
        # Update particle position
        self.positions[i] = [self.positions[i][j] + self.velocities[i][j] for j in range(self.D)]
        # Ensure positions stay within bounds
        self.positions[i] = [max(self.lower_bound, min(self.upper_bound, x)) for x in self.positions[i]]
        
    def update_personal_best(self, i):
        # Update personal best if the current position is better
        current_value = self.cost_function(self.positions[i])
        if current_value < self.p_best_values[i]:
            self.p_best[i] = list(self.positions[i])
            self.p_best_values[i] = current_value

    def update_global_best(self):
        # Update global best position
        best_particle_idx = self.p_best_values.index(min(self.p_best_values))
        self.g_best = list(self.p_best[best_particle_idx])
        
    def optimize(self):
        for t in range(self.Nc):
            for i in range(self.Np):
                # Update the velocity, position, and personal best of each particle
                self.update_velocity(i, t)
                self.update_position(i)
                self.update_personal_best(i)
            # Update global best
            self.update_global_best()

            # Store the global best cost for plotting
            self.global_best_costs.append(self.cost_function(self.g_best))
            
            # Print the global best cost and position for each iteration
            print(f"Cycle {t+1}: Global Best Cost = {self.cost_function(self.g_best)}, Global Best Position = {self.g_best}")
            
        return self.g_best

# Parameters
Np = 30  # Number of particles
Nc = 100  # Maximum number of cycles
D = 5  # Dimension of position and velocity
alpha = 0.9  # Inertia weight
beta = 1.5  # Acceleration constant
lower_bound = -5  # Lower bound for particle positions
upper_bound = 5  # Upper bound for particle positions

# Initialize and run the PSO algorithm
pso = PSO(cost_function, Np, Nc, D, alpha, beta, lower_bound, upper_bound)
best_position = pso.optimize()

# Plot the global best cost over cycles
plt.plot(range(Nc), pso.global_best_costs)
plt.xlabel('Cycle')
plt.ylabel('Global Best Cost')
plt.title('PSO Optimization Progress')
plt.show()

# Print the final global best position
print(f"Final Global Best Position: {best_position}")
````

### Genetic Learning Particle Swarm Optimization (GL-PSO) Algorithm

GL-PSO is an optimization algorithm that combines genetic learning with particle swarm optimization in a highly cohesive way. The algorithm is summarized as follows:

#### **Algorithm: GL-PSO**

1. **Initialize**:
    - For $ i = 1 $ to $ M $:
        1. Randomly initialize $ v_i $ and $ x_i $;
        2. Evaluate $ f(x_i) $;
        3. Set $ p_{i,\text{best}} = x_i $.
    - End For.
    - Set $ g_{\text{best}} $ to the current best position among particles.

2. **Repeat**:
    - For $ i = 1 $ to $ M $:
        1. For $ d = 1 $ to $ D $:
            - Randomly select a particle $ k \in \{1, \dots, M\} $;
            - If $ f(p_{i,\text{best}}) < f(p_{k,\text{best}}) $:
                $$
                o_{i,d} = r_d \cdot p_{i,d} + (1 - r_d) \cdot g_d
                $$
            - Else:
                $$
                o_{i,d} = p_{k,d}
                $$
            - End If.
        - End For.
        2. For $ d = 1 $ to $ D $:
            - If $ \text{rand}(0, 1) < p_m $:
                $$
                o_{i,d} = \text{rand}(lbd, ubd)
                $$
            - End If.
        - End For.
        3. Evaluate $ f(O_i) $.
        4. If $ f(O_i) < f(e_i) $:
            $$
            e_i = O_i
            $$
        - End If.
        5. If $ f(e_i) $ ceases improving for $ sg $ generations:
            - Select $ e_j $ by $ 20\%\cdot M $ tournament;
            $$
            e_i = e_j
            $$
        - End If.
        6. For $ d = 1 $ to $ D $:
            - Update velocity:
            $$
            v_{i,d} = \omega \cdot v_{i,d} + c \cdot r_d \cdot (e_{i,d} - x_{i,d})
            $$
            - Update position:
            $$
            x_{i,d} = x_{i,d} + v_{i,d}
            $$
        - End For.
        7. Evaluate $ f(X_i) $.
        8. Update $ P_i $ and $ G $.
    - End For.
3. **Until** Terminal Condition.

---

### **Advantages of PSO over GA**

- PSO is easier to implement with fewer parameters to adjust.
- PSO has a more effective memory capability:
    - Each particle remembers its own previous best $( p_{\text{best}} $) and the neighborhood/global best $( g_{\text{best}} $).
- PSO maintains diversity in the swarm better than GA:
    - All particles use the information of the most successful particle.
    - In GA, worse solutions are discarded, and the population evolves around a subset of best individuals, reducing diversity.

---

### **Drawbacks of PSO**

1. **Premature Convergence**:
    - For global best PSO, particles may converge to a single point on the line between $ g_{\text{best}} $ and $ p_{\text{best}} $, which is not guaranteed to be a local optimum.
    - Fast information flow may result in loss of diversity, increasing the likelihood of being trapped in local optima.
    
2. **Problem-Dependent Performance**:
    - Stochastic approaches like PSO are sensitive to parameter settings.
    - **Inertia Weight**:
        - Increasing inertia weight promotes global search (exploration).
        - Decreasing inertia weight promotes local search (exploitation).

---

### **Comparison Between PSO and GA**

| **Aspect**                 | **PSO**                                                         | **GA**                                   |
|----------------------------|----------------------------------------------------------------|-----------------------------------------|
| **Search Strategy**        | Particles guided by $ p_{\text{best}} $ and $ g_{\text{best}} $ | Uses crossover and mutation for exploration |
| **Convergence**            | Faster convergence                                             | Better exploration                      |
| **Diversity Maintenance**  | Maintains diversity through global best sharing                | Limited due to selection of best subset |
| **Hybrid Potential**       | Can include mutation for increased diversity                   | Swapping information via crossover      |

---

### **Hybrid GA-PSO and GL-PSO**

- **GA-PSO**: Combines genetic algorithms with particle swarm optimization:
    - Incorporates advantages of swarm intelligence (PSO) and natural selection (GA).
    - Increases diversity and avoids local maxima by introducing mutation.
    
- **GL-PSO**: A more cohesive hybridization of genetic learning and PSO:
    - Genetic learning guides particle updates.
    - Retains the advantages of both approaches for improved optimization performance.


````python
import random
import matplotlib.pyplot as plt

# Define the cost function (example: Sphere function)
def cost_function(x):
    return sum([xi ** 2 for xi in x])

# Genetic Learning Particle Swarm Optimization (GL-PSO) class
class GLPSO:
    def __init__(self, cost_function, M, D, max_gen, omega, c, pm, lb, ub, sg):
        self.cost_function = cost_function
        self.M = M  # Number of particles
        self.D = D  # Dimensionality
        self.max_gen = max_gen  # Maximum generations
        self.omega = omega  # Inertia weight
        self.c = c  # Acceleration coefficient
        self.pm = pm  # Mutation probability
        self.lb = lb  # Lower bound
        self.ub = ub  # Upper bound
        self.sg = sg  # Stagnation generations for resetting
        self.positions = [[random.uniform(lb, ub) for _ in range(D)] for _ in range(M)]
        self.velocities = [[random.uniform(-1, 1) for _ in range(D)] for _ in range(M)]
        self.p_best = [list(pos) for pos in self.positions]
        self.p_best_cost = [self.cost_function(pos) for pos in self.positions]
        self.g_best = min(self.p_best, key=self.cost_function)
        self.g_best_cost = self.cost_function(self.g_best)
        self.global_costs = []

    def optimize(self):
        stagnation_counter = [0] * self.M  # Tracks stagnation for each particle

        for gen in range(self.max_gen):
            for i in range(self.M):
                offspring = []
                for d in range(self.D):
                    k = random.randint(0, self.M - 1)  # Randomly select a particle
                    if self.p_best_cost[i] < self.p_best_cost[k]:
                        r_d = random.random()
                        offspring.append(r_d * self.p_best[i][d] + (1 - r_d) * self.g_best[d])
                    else:
                        offspring.append(self.p_best[k][d])

                # Mutation step
                offspring = [
                    random.uniform(self.lb, self.ub) if random.random() < self.pm else o_d
                    for o_d in offspring
                ]

                # Evaluate offspring
                offspring_cost = self.cost_function(offspring)
                if offspring_cost < self.p_best_cost[i]:
                    self.p_best[i] = list(offspring)
                    self.p_best_cost[i] = offspring_cost

                # Check stagnation and reset if needed
                if stagnation_counter[i] >= self.sg:
                    tournament = random.sample(range(self.M), max(1, self.M // 5))
                    best_tournament_particle = min(tournament, key=lambda idx: self.p_best_cost[idx])
                    self.p_best[i] = list(self.p_best[best_tournament_particle])
                    stagnation_counter[i] = 0
                else:
                    stagnation_counter[i] += 1

                # Update velocity and position
                for d in range(self.D):
                    r_d = random.random()
                    self.velocities[i][d] = (
                        self.omega * self.velocities[i][d]
                        + self.c * r_d * (self.p_best[i][d] - self.positions[i][d])
                    )
                    self.positions[i][d] += self.velocities[i][d]
                    # Keep positions within bounds
                    self.positions[i][d] = max(self.lb, min(self.ub, self.positions[i][d]))

            # Update global best
            current_best_particle = min(range(self.M), key=lambda idx: self.p_best_cost[idx])
            if self.p_best_cost[current_best_particle] < self.g_best_cost:
                self.g_best = list(self.p_best[current_best_particle])
                self.g_best_cost = self.p_best_cost[current_best_particle]

            # Store global best cost for plotting
            self.global_costs.append(self.g_best_cost)
            print(f"Generation {gen + 1}: Global Best Cost = {self.g_best_cost}")

        return self.g_best, self.g_best_cost

# Parameters
M = 30  # Number of particles
D = 5  # Dimension of the search space
max_gen = 100  # Maximum number of generations
omega = 0.7  # Inertia weight
c = 1.5  # Acceleration coefficient
pm = 0.1  # Mutation probability
lb = -10  # Lower bound
ub = 10  # Upper bound
sg = 10  # Stagnation threshold

# Initialize and run the GL-PSO
gl_pso = GLPSO(cost_function, M, D, max_gen, omega, c, pm, lb, ub, sg)
best_position, best_cost = gl_pso.optimize()

# Plot the global best cost over generations
plt.plot(range(1, max_gen + 1), gl_pso.global_costs)
plt.xlabel('Generation')
plt.ylabel('Global Best Cost')
plt.title('GL-PSO Optimization Progress')
plt.show()

# Print the final best solution
print(f"Final Global Best Position: {best_position}")
print(f"Final Global Best Cost: {best_cost}")
````

## Hybrid Particle Swarm Optimization with Local Search (HPSO-LS)

This notebook implements the HPSO-LS algorithm for feature selection in supervised classification. The algorithm combines Particle Swarm Optimization (PSO) with local search techniques to improve exploration and fine-tuning near local optima.

---

## Problem Description

Given a dataset:
-$ S = \{(x_1, y_1), \ldots, (x_n, y_n)\} $
  - $ x_i = [x_{i1}, \ldots, x_{id}]^T $: Multi-dimensional vector sample.
  - $ d $: Number of features.
  - $ n $: Number of samples.
  - $ y_i $: Label of sample $ x_i $.

The goal is to select an optimal subset of features $ X_s \subseteq X $ to improve model performance while reducing dimensionality.

---

## HPSO-LS Algorithm

The pseudo-code for HPSO-LS is provided below:

### Algorithm : HPSO-LS

1. **Input**: $ \text{Feature} = \{f_1, \ldots, f_D\} $, $ N_C^{\text{max}} $: maximum iterations, $ N_P $: number of particles.
2. **Initialize**:
    - For $ i = 1 $ to $ N_P $:
        - Randomly initialize $ x_{ij} \in \{0, 1\} $ and $ v_{ij} \sim U[0, 1] $.
    - Compute pairwise correlation coefficients:
      $$
      c_{ij} = \frac{\sum_{k=1}^m (x_i(k) - \bar{x}_i)(x_j(k) - \bar{x}_j)}{\sqrt{\sum_{k=1}^m (x_i(k) - \bar{x}_i)^2} \cdot \sqrt{\sum_{k=1}^m (x_j(k) - \bar{x}_j)^2}}
      $$
    - Compute:
      $$
      \text{cor}_i = \frac{1}{n_f - 1} \sum_{j=1, j \neq i}^{n_f} |c_{ij}|
      $$
    - Partition features:
        - $ \text{Similar features} $: $ \{f_i | \text{cor}_i \geq \text{cor}_{\text{mid}}\} $.
        - $ \text{Dissimilar features} $: $ \{f_i | \text{cor}_i < \text{cor}_{\text{mid}}\} $.
3. **Update velocities and positions**:
    - Update velocity:
      $$
      \hat{v}_i(t+1) = v_i(t) + c_1 r_1 (\text{rand} \cdot (x_i^{\text{best}} - x_i(t))) + c_2 r_2 (\text{rand} \cdot (x_g^{\text{best}} - x_i(t)))
      $$
      $$
      v_{id}(t+1) = \text{sign}(\hat{v}_{id}(t+1)) \cdot \min(|\hat{v}_{id}(t+1)|, v_{\text{max}})
      $$
    - Update position:
      $$
      x_{id}(t+1) =
      \begin{cases} 
      1 & \text{if } \text{rand} < \frac{1}{1 + \exp(-v_{id}(t+1))} \\
      0 & \text{otherwise}
      \end{cases}
      $$
4. **Refine features**:
    - Update $ X_s $: Similar features selected by the particle.
    - Update $ X_d $: Dissimilar features selected by the particle.
    - Remove features from $ X_s $ and $ X_d $ with $ x_{id}(t+1) = 0 $.
    - Compute fitness $ f_i = \text{fitness}(X_s) $.
5. **Evaluate fitness**:
    - Update personal and global bests:
      $$
      f_i^{\text{best}} =
      \begin{cases}
      f(X_i) & \text{if } f(X_i) > f_i^{\text{best}} \\
      f_g^{\text{best}} & \text{if } f(X_i) > f_g^{\text{best}} \\
      f_i^{\text{best}} & \text{otherwise}
      \end{cases}
      $$
6. **Output**: Selected features $ \text{Feature} = \{f_1, \ldots, f_k\} $.

````python
## Implementation

# Below is the Python implementation of HPSO-LS:

# ```python
import random
import math

# Fitness function (example: classification accuracy)
def fitness(selected_features):
    # Placeholder for actual model evaluation
    return sum(selected_features) / len(selected_features)

# HPSO-LS algorithm
class HPSO_LS:
    def __init__(self, n_features, n_particles, max_cycles, v_max, c1, c2, cor_mid):
        self.n_features = n_features
        self.n_particles = n_particles
        self.max_cycles = max_cycles
        self.v_max = v_max
        self.c1 = c1
        self.c2 = c2
        self.cor_mid = cor_mid
        self.particles = [
            {"position": [random.randint(0, 1) for _ in range(n_features)],
             "velocity": [random.uniform(0, 1) for _ in range(n_features)]}
            for _ in range(n_particles)
        ]
        self.g_best = None
        self.g_best_fitness = -float("inf")

    def optimize(self):
        for cycle in range(self.max_cycles):
            for particle in self.particles:
                fitness_value = fitness(particle["position"])
                if fitness_value > self.g_best_fitness:
                    self.g_best = particle["position"]
                    self.g_best_fitness = fitness_value

                for d in range(self.n_features):
                    r1, r2 = random.random(), random.random()
                    particle["velocity"][d] = (
                        particle["velocity"][d]
                        + self.c1 * r1 * (self.g_best[d] - particle["position"][d])
                        + self.c2 * r2 * (self.g_best[d] - particle["position"][d])
                    )
                    particle["velocity"][d] = max(
                        -self.v_max, min(self.v_max, particle["velocity"][d])
                    )
                    sigmoid = 1 / (1 + math.exp(-particle["velocity"][d]))
                    particle["position"][d] = 1 if random.random() < sigmoid else 0
            print(f"Cycle {cycle + 1}: Best Fitness = {self.g_best_fitness}")

        return self.g_best, self.g_best_fitness

# Parameters
n_features = 10
n_particles = 30
max_cycles = 100
v_max = 4
c1, c2 = 1.5, 1.5
cor_mid = 0.5

# Run HPSO-LS
hpso_ls = HPSO_LS(n_features, n_particles, max_cycles, v_max, c1, c2, cor_mid)
best_features, best_fitness = hpso_ls.optimize()

print("Selected Features:", best_features)
print("Best Fitness:", best_fitness)
````

## Opposition-Based Evolutionary Computation

---

## Introduction

Opposition-Based Evolutionary Computation (OBEC) improves machine intelligence algorithms by simultaneously considering candidate solutions and their opposites. This approach accelerates convergence by increasing the likelihood of starting near the optimal solution.

Key concepts include:

- **Opposition-Based Learning (OBL)**: Evaluates both the current solution and its opposite.
- **Opposition-Based Optimization (OBO)**: Incorporates opposition solutions into optimization algorithms.

---

## Definitions

### Opposite Number
For a real number $ x $ in the interval $[a, b]$, the opposite number is defined as:
$$
\bar{x} = (a + b) - x
$$

### Opposite Point
For a vector $ \mathbf{x} = [x_1, x_2, \ldots, x_D]^T $ in $ D $-dimensional space with $ x_i \in [a_i, b_i] $, the opposite point is:
$$
\bar{x}_i = a_i + b_i - x_i, \quad i = 1, 2, \ldots, D
$$

### Generalized Opposition Solution
A generalized opposition solution introduces a scaling factor $ k \in [0, 1] $:
$$
\tilde{x}_i = k \cdot (a_i + b_i) - x_i, \quad i = 1, 2, \ldots, D
$$

### Center-Based Sampling (CBS)
A random point between a solution and its opposite:
$$
\hat{x}_i = \text{rand}_i \cdot (a_i + b_i - 2x_i) + x_i, \quad \text{rand}_i \in [0, 1]
$$

---

## Opposition-Based Learning (OBL)

### Definition
Given:
- Solution point $ \mathbf{x} = [x_1, \ldots, x_D]^T $
- Opposite solution point $ \bar{\mathbf{x}} = [\bar{x}_1, \ldots, \bar{x}_D]^T $

If the opposite solution $ \bar{\mathbf{x}} $ has better fitness:
$$
f(\bar{\mathbf{x}}) < f(\mathbf{x})
$$
then replace $ \mathbf{x} $ with $ \bar{\mathbf{x}} $. Otherwise, keep $ \mathbf{x} $.

---

## Algorithm Steps for OBL in Evolutionary Computation

1. **Initialization**:
   - Generate initial population of solutions $ \mathbf{x}_i $.
   - Compute the opposite solutions $ \bar{\mathbf{x}}_i $.

2. **Fitness Evaluation**:
   - Evaluate the fitness of both $ \mathbf{x}_i $ and $ \bar{\mathbf{x}}_i $.

3. **Selection**:
   - Replace $ \mathbf{x}_i $ with $ \bar{\mathbf{x}}_i $ if $ f(\bar{\mathbf{x}}_i) < f(\mathbf{x}_i) $.

4. **Evolution**:
   - Apply evolutionary operators (e.g., crossover, mutation).

5. **Iteration**:
   - Repeat steps 2–4 until convergence or maximum iterations.

---

## Applications

OBL has been successfully integrated into several optimization algorithms, including:

- **Genetic Algorithms (GA)**
- **Differential Evolution (DE)**
- **Particle Swarm Optimization (PSO)**
- **Harmony Search (HS)**

### Advantages
- Faster convergence by leveraging opposite solutions.
- Improved exploration of the search space.
- Effective for high-dimensional optimization problems.

````python
import numpy as np
import random

# Define a simple fitness function (e.g., sphere function)
def fitness(x):
    return np.sum(x**2)

# Generate the opposite solution of a given solution x
def get_opposite_solution(x, lower_bound, upper_bound):
    return lower_bound + upper_bound - x

# Create a random solution
def create_random_solution(dim, lower_bound, upper_bound):
    return np.random.uniform(lower_bound, upper_bound, dim)

# Crossover function (e.g., single point crossover)
def crossover(parent1, parent2):
    crossover_point = random.randint(0, len(parent1) - 1)
    offspring1 = np.copy(parent1)
    offspring2 = np.copy(parent2)
    
    offspring1[crossover_point:] = parent2[crossover_point:]
    offspring2[crossover_point:] = parent1[crossover_point:]
    
    return offspring1, offspring2

# Mutation function (e.g., random mutation)
def mutate(solution, lower_bound, upper_bound, mutation_rate=0.1):
    for i in range(len(solution)):
        if random.random() < mutation_rate:
            solution[i] = np.random.uniform(lower_bound, upper_bound)
    return solution

# Main function for Opposition-Based Evolutionary Computation
def opposition_based_evolutionary_computation(pop_size, dim, lower_bound, upper_bound, max_generations):
    # Initialize population
    population = [create_random_solution(dim, lower_bound, upper_bound) for _ in range(pop_size)]
    best_solution = None
    best_fitness = float('inf')

    for generation in range(max_generations):
        new_population = []

        for i in range(0, pop_size, 2):
            # Select two random parents
            parent1 = population[i]
            parent2 = population[i + 1] if i + 1 < pop_size else population[0]  # Wrap around if odd number of individuals
            
            # Create offspring via crossover
            offspring1, offspring2 = crossover(parent1, parent2)
            
            # Mutate the offspring
            offspring1 = mutate(offspring1, lower_bound, upper_bound)
            offspring2 = mutate(offspring2, lower_bound, upper_bound)
            
            # Get the opposite solutions
            opposite1 = get_opposite_solution(offspring1, lower_bound, upper_bound)
            opposite2 = get_opposite_solution(offspring2, lower_bound, upper_bound)
            
            # Evaluate fitness of offspring and their opposites
            fitness1 = fitness(offspring1)
            fitness2 = fitness(offspring2)
            opposite_fitness1 = fitness(opposite1)
            opposite_fitness2 = fitness(opposite2)
            
            # Replace with the better solution (original or opposite)
            if opposite_fitness1 < fitness1:
                offspring1 = opposite1
            if opposite_fitness2 < fitness2:
                offspring2 = opposite2
            
            new_population.extend([offspring1, offspring2])
        
        # Evaluate the best solution in the population
        for individual in new_population:
            current_fitness = fitness(individual)
            if current_fitness < best_fitness:
                best_fitness = current_fitness
                best_solution = individual

        # Replace the old population with the new one
        population = new_population

        # Print the best fitness in the current generation
        print(f"Generation {generation+1}, Best Fitness: {best_fitness}")

    return best_solution, best_fitness

# Parameters
pop_size = 20  # Population size
dim = 5        # Dimensions of the solution
lower_bound = -5
upper_bound = 5
max_generations = 100  # Max generations

# Run the algorithm
best_solution, best_fitness = opposition_based_evolutionary_computation(pop_size, dim, lower_bound, upper_bound, max_generations)
print(f"Best Solution: {best_solution}")
print(f"Best Fitness: {best_fitness}")
````

## Opposition-Based Differential Evolution (ODE)

In **Evolutionary Optimization**, the process starts with an initial population that is randomly generated. The goal is to improve the candidate solutions iteratively to find the optimal solution. One way to enhance the evolutionary optimization process is by utilizing **Opposition-Based Learning (OBL)**. This can help in improving the convergence rate by simultaneously checking both a candidate solution and its opposite solution.

###  Opposition-Based Differential Evolution (ODE)

In traditional **Differential Evolution (DE)**, the optimization process starts with randomly generated solutions, and the evolutionary process improves these solutions by applying mutation and crossover operations. However, if the initial guesses are far from the optimal solution, the convergence might take longer. 

Opposition-Based Differential Evolution (ODE) enhances the classical DE algorithm by incorporating opposition-based techniques at two key stages:

1. **Opposition-Based Population Initialization**: This step involves creating an initial population, and then calculating the opposite population. The fittest individuals from the combined set of original and opposite populations are selected as the initial population.

2. **Opposition-Based Generation Jumping**: This step applies a generation jumping mechanism, where the opposite solutions are created by considering the minimum and maximum values of each dimension in the population. These opposite solutions are then evaluated, and the best individuals are selected to form the new population.

### Algorithm : Opposition-Based Differential Evolution (ODE)

The **ODE** algorithm consists of the following steps:

1. **Input Parameters**:
   - Population size $ N_p $
   - Problem dimension $ D $
   - Mutation constant $ F $
   - Crossover rate $ C_r $

2. **Opposition-Based Population Initialization**:
   - Generate a uniformly distributed random initial population $ P_0 $
   - Calculate the opposite population $ OP_0 $ as:
     $$
     OP_{0_{i,j}} = a_j + b_j - P_{0_{i,j}}, \quad \text{for } i = 1, \dots, N_p \quad \text{and} \quad j = 1, \dots, D
     $$
   - Select the $ N_p $ fittest individuals from the combined set $ \{P_0, OP_0\} $ to form the initial population $ P_0 $.

3. **Evolutionary Steps (Differential Evolution)**:
   - Apply the standard Differential Evolution (DE) mutation, crossover, and selection steps to generate new candidates.

4. **Opposition-Based Generation Jumping**:
   - With a probability $ J_r $, create opposite solutions $ OP $ from the current population using:
     $$
     OP_{i,j} = MIN_j + MAX_j - P_{i,j}, \quad \text{for } i = 1, \dots, N_p \quad \text{and} \quad j = 1, \dots, D
     $$
   - Select the $ N_p $ fittest individuals from the combined set $ \{P, OP\} $ as the new current population $ P $.

5. **Termination**:
   - Repeat the evolutionary steps and generation jumping until the stopping criterion is met (e.g., maximum number of generations or convergence threshold).

### Steps in ODE Algorithm

$$
\text{Algorithm : Opposition-Based Differential Evolution (ODE)}
$$

$$
\text{Input: } N_p, D, F, C_r \quad \text{(population size, problem dimension, mutation constant, crossover rate)}
$$
$$
\text{Step 1: Initialize the population } P_0 \quad \text{(random)}
$$
$$
\text{Step 2: Calculate opposite population } OP_0
$$
$$
OP_{0_{i,j}} = a_j + b_j - P_{0_{i,j}} \quad \text{for } i = 1, \dots, N_p \quad \text{and} \quad j = 1, \dots, D
$$
$$
\text{Step 3: Select the fittest individuals from } \{P_0, OP_0\} \quad \text{(initial population)}
$$
$$
\text{Step 4: Perform Differential Evolution (DE) steps}
$$
$$
\text{Step 5: Opposition-Based Generation Jumping:}
$$
$$
\text{If } rand(0, 1) < J_r \text{, then:}
$$
$$
OP_{i,j} = MIN_j + MAX_j - P_{i,j} \quad \text{for } i = 1, \dots, N_p \quad \text{and} \quad j = 1, \dots, D
$$
$$
\text{Step 6: Select the fittest individuals from } \{P, OP\} \quad \text{(new population)}
$$
$$
\text{Step 7: Repeat until stopping criterion is met}
$$

---

The opposition-based initialization and generation jumping are the key features that differentiate ODE from standard DE. These techniques aim to improve convergence by generating more promising solutions earlier in the optimization process.

### Conclusion

Opposition-Based Differential Evolution (ODE) enhances the classical Differential Evolution algorithm by leveraging opposition-based learning techniques. By using both original and opposite solutions for population initialization and generation jumping, ODE aims to reduce the distance between the population and the optimal solution, leading to faster convergence and better optimization performance.


````python
import random

def fitness_function(x):
    # Example fitness function: Sphere function (sum of squares)
    return sum([xi**2 for xi in x])

def initialize_population(pop_size, dim, lower_bound, upper_bound):
    population = []
    for _ in range(pop_size):
        individual = [random.uniform(lower_bound, upper_bound) for _ in range(dim)]
        population.append(individual)
    return population

def opposition_based_initialization(population, lower_bound, upper_bound):
    opposite_population = []
    for individual in population:
        opposite_individual = [lower_bound + upper_bound - xi for xi in individual]
        opposite_population.append(opposite_individual)
    return opposite_population

def select_fittest(population, opposite_population, fitness_func):
    combined_population = population + opposite_population
    fitness_values = [fitness_func(individual) for individual in combined_population]
    
    # Sort by fitness values, selecting the best N individuals
    sorted_indices = sorted(range(len(fitness_values)), key=lambda k: fitness_values[k])
    best_indices = sorted_indices[:len(population)]
    
    selected_population = [combined_population[i] for i in best_indices]
    
    return selected_population

def differential_evolution(population, fitness_func, mutation_factor=0.8, crossover_rate=0.7, lower_bound=-5.12, upper_bound=5.12):
    # Apply mutation, crossover, and selection for evolution
    pop_size = len(population)
    dim = len(population[0])
    
    for i in range(pop_size):
        # Mutation: generate a mutant vector
        indices = list(range(pop_size))
        indices.remove(i)
        a, b, c = random.sample(indices, 3)
        
        mutant = [population[a][j] + mutation_factor * (population[b][j] - population[c][j]) for j in range(dim)]
        
        # Crossover: create trial vector by combining mutant and current individual
        trial = []
        for j in range(dim):
            if random.random() < crossover_rate:
                trial.append(mutant[j])
            else:
                trial.append(population[i][j])
        
        # Selection: choose the better of the current individual or the trial vector
        if fitness_func(trial) < fitness_func(population[i]):
            population[i] = trial
            
    return population

def opposition_based_generation_jumping(population, fitness_func, lower_bound=-5.12, upper_bound=5.12):
    # Generate opposite solutions based on min/max values of the population
    pop_size = len(population)
    dim = len(population[0])
    
    # Get min and max values for each dimension in the population
    min_vals = [min([individual[j] for individual in population]) for j in range(dim)]
    max_vals = [max([individual[j] for individual in population]) for j in range(dim)]
    
    opposite_population = []
    for individual in population:
        opposite_individual = [min_vals[j] + max_vals[j] - individual[j] for j in range(dim)]
        opposite_population.append(opposite_individual)
    
    # Select the fittest individuals from the current population and the opposite population
    population = select_fittest(population, opposite_population, fitness_func)
    
    return population

def opposition_based_differential_evolution(pop_size, dim, lower_bound, upper_bound, max_generations, mutation_factor=0.8, crossover_rate=0.7):
    # Step 1: Initialize population
    population = initialize_population(pop_size, dim, lower_bound, upper_bound)
    
    # Step 2: Opposition-based initialization
    opposite_population = opposition_based_initialization(population, lower_bound, upper_bound)
    population = select_fittest(population, opposite_population, fitness_function)
    
    # Step 3: Evolutionary process with opposition-based generation jumping
    for generation in range(max_generations):
        # Apply Differential Evolution (mutation, crossover, and selection)
        population = differential_evolution(population, fitness_function, mutation_factor, crossover_rate, lower_bound, upper_bound)
        
        # Apply Opposition-Based Generation Jumping with probability
        if random.random() < 0.1:  # Example probability for generation jumping
            population = opposition_based_generation_jumping(population, fitness_function, lower_bound, upper_bound)
        
        # Optionally, print progress or evaluate stopping criteria
        if generation % 10 == 0:
            best_fitness = min([fitness_function(individual) for individual in population])
            print(f"Generation {generation}: Best Fitness = {best_fitness}")
    
    # Return the final population
    return population

# Example usage
pop_size = 50
dim = 30
lower_bound = -5.12
upper_bound = 5.12
max_generations = 100

# Run the Opposition-Based Differential Evolution algorithm
final_population = opposition_based_differential_evolution(pop_size, dim, lower_bound, upper_bound, max_generations)

# Print the best solution found
best_solution = min(final_population, key=fitness_function)
best_fitness = fitness_function(best_solution)
print("Best Solution:", best_solution)
print("Best Fitness:", best_fitness)
````

###  Two Variants of Opposition-Based Learning

In opposition-based learning, an opposite point is deterministically calculated with the reference point at the center in the fixed search range. There are two variants of opposite point selection. The first variant is called **quasi opposition-based learning (quasi-OBL)**, proposed in [130].

#### Quasi-Opposite Number

Let $ x $ be any real number between $ [a, b] $ and $ x_o = \hat{x} $ be its opposite number. The **quasi-opposite number** of $ x $, denoted by $ x^q $, is defined as:

$$
x^q = \text{rand}(m, x_o),
$$

where $ m = \frac{a + b}{2} $ is the middle of the interval $ [a, b] $ and $ \text{rand}(m, x_o) $ is a random number uniformly distributed between $ m $ and the opposite point $ x_o $.

#### Quasi-Opposite Point

Let $ \mathbf{x}_i = [x_{i1}, \dots, x_{iD}]^T \in \mathbb{R}^D $ be any vector (point) in $ D $-dimensional real space, where $ x_{ij} \in [a_j, b_j] $ for $ i = 1, \dots, N_p $ and $ j = 1, \dots, D $. The **quasi-opposite point** of $ \mathbf{x}_i $, denoted by $ \mathbf{x}_i^q $, is defined in its element form as:

$$
x_{ij}^q = \text{rand}(M_{ij}, x_{ij}^o),
$$

where $ M_{ij} = \frac{a_j + b_j}{2} $ is the middle of the interval $ [a_j, b_j] $, and $ \text{rand}(M_{ij}, x_{ij}^o) $ is a random number uniformly distributed between the middle $ M_{ij} $ and the opposite number $ x_{ij}^o $. The calculation of the quasi-opposite point depends on the comparison of the values between the opposite point and the original point as follows:

$$
x_{ij} = 
\begin{cases}
M_{ij} + \text{rand}(0, 1) \cdot (x_{ij}^o - M_{ij}), & \text{if } x_{ij} < M_{ij}, \\
x_{ij}^o + \text{rand}(0, 1) \cdot (M_{ij} - x_{ij}^o), & \text{otherwise}.
\end{cases}
$$

#### Theorem  (Quasi-Opposite Point Selection)

Given a guess point $ \mathbf{x} \in \mathbb{R}^D $, its opposite point $ \mathbf{x}^o \in \mathbb{R}^D $ and quasi-opposite point $ \mathbf{x}^q \in \mathbb{R}^D $, and given the distance from the solution $ d(\cdot) $ and probability function $ \text{Pr}(\cdot) $, then:

$$
\text{Pr}[d(\mathbf{x}^q) < d(\mathbf{x}^o)] > \frac{1}{2}.
$$


````python
import random

# Define a function for generating the quasi-opposite point
def quasi_opposite_point(x, a, b):
    """
    Generate a quasi-opposite point for a given point `x` in the interval [a, b].
    Parameters:
        x (list): The current point in the search space (D-dimensional vector).
        a (list): Lower bounds for each dimension of the search space.
        b (list): Upper bounds for each dimension of the search space.
    Returns:
        list: The quasi-opposite point.
    """
    D = len(x)  # Number of dimensions
    quasi_opposite = []
    
    for i in range(D):
        # Compute the middle of the interval for the i-th dimension
        Mij = (a[i] + b[i]) / 2
        # Calculate the opposite point for the i-th dimension
        xij_o = a[i] + b[i] - x[i]
        
        # Generate the quasi-opposite point
        if x[i] < Mij:
            # Generate a random number between Mij and xij_o
            quasi_opposite.append(Mij + random.random() * (xij_o - Mij))
        else:
            # Generate a random number between xij_o and Mij
            quasi_opposite.append(xij_o + random.random() * (Mij - xij_o))
    
    return quasi_opposite

# Example usage
a = [0, 0]  # Lower bounds for each dimension
b = [10, 10]  # Upper bounds for each dimension

# Example solution point
x = [3, 7]

# Generate the quasi-opposite point for the solution `x`
qop = quasi_opposite_point(x, a, b)

print("Original Point:", x)
print("Quasi-Opposite Point:", qop)
````

# Evolutionary Computation: An Overview

Evolutionary computation is a subset of artificial intelligence, which is essentially a computational intelligence for solving combinatorial optimization problems. It starts with a group of randomly generated individuals and mimics the biological genetic mode to update the next generation of individuals through replication, crossover, and mutation.

Then, based on fitness size, the individual survival of the fittest improves the quality of the new generation of individuals. After repeated iterations, the algorithm gradually approaches the optimal solution. From a mathematical point of view, evolutionary computation is essentially a search optimization method.

## Key Concepts in Evolutionary Computation:

1. **Initial Population**: Starting from a group of randomly generated individuals.
2. **Evolutionary Operations**: Replication, crossover, mutation.
3. **Survival of the Fittest**: The individual with better fitness is selected to form the next generation.
4. **Optimization**: Evolutionary computation is used as a search optimization method.

## Pareto Optimization and Coevolutionary Computation

This chapter also focuses on **Pareto optimization theory** in multi-objective optimization problems that arise in **coevolutionary computation**.

## Applications of Evolutionary Computation

Evolutionary computation has been widely used in various fields including:
- Artificial Intelligence
- Pattern Recognition
- Image Processing
- Biology
- Electrical Engineering
- Communication
- Economic Management
- Mechanical Engineering

## Quasi-Opposition-Based Learning (Quasi-OBL)

After calculating the fitness of the opposite population generated by **quasi-reflected opposition**, the fittest $ N_p $ individuals from the populations $ P $ and $ OP $ constitute the updated population. 

The application of the **quasi-opposite points** or **quasi-reflected opposite points** will result in various variants and extensions of the basic OBL algorithm. For example, a **stochastic opposition-based learning** using a **Beta distribution** in differential evolution can be found in [121].



````python
import random

def fitness_function(x):
    # Example fitness function: Sphere function (minimize)
    return sum([xi**2 for xi in x])

def create_population(pop_size, dim, lower_bound, upper_bound):
    """Create a random population."""
    return [[random.uniform(lower_bound, upper_bound) for _ in range(dim)] for _ in range(pop_size)]

def quasi_opposite_point(x, lower_bound, upper_bound):
    """Calculate the quasi-opposite point for a solution x."""
    dim = len(x)
    quasi_opposite = []
    
    for i in range(dim):
        m_ij = (lower_bound + upper_bound) / 2
        op_ij = lower_bound + upper_bound - x[i]
        
        if x[i] < m_ij:
            quasi_opposite.append(m_ij + random.random() * (op_ij - m_ij))
        else:
            quasi_opposite.append(op_ij + random.random() * (m_ij - op_ij))
    
    return quasi_opposite

def differential_evolution(pop_size, dim, lower_bound, upper_bound, max_gen, F, Cr):
    """Differential Evolution Algorithm with Quasi-OBL."""
    # Initialize population
    population = create_population(pop_size, dim, lower_bound, upper_bound)
    
    for generation in range(max_gen):
        new_population = []
        
        # Apply Quasi-OBL to improve the population
        for i in range(pop_size):
            x_i = population[i]
            op_i = quasi_opposite_point(x_i, lower_bound, upper_bound)
            
            # Compare fitness of current solution and its quasi-opposite
            if fitness_function(op_i) < fitness_function(x_i):
                new_population.append(op_i)
            else:
                new_population.append(x_i)
        
        # Differential evolution mutation and crossover (with some adjustments)
        for i in range(pop_size):
            # Mutation
            idxs = list(range(pop_size))
            idxs.remove(i)  # Remove self
            a, b, c = random.sample(idxs, 3)
            donor_vector = [population[a][j] + F * (population[b][j] - population[c][j]) for j in range(dim)]
            
            # Crossover
            trial_vector = [
                donor_vector[j] if random.random() < Cr else population[i][j] 
                for j in range(dim)
            ]
            
            # Selection (choose the better of trial or current solution)
            if fitness_function(trial_vector) < fitness_function(population[i]):
                new_population[i] = trial_vector
        
        population = new_population  # Update population
        
        # Output the best solution of each generation
        best_solution = min(population, key=fitness_function)
        print(f"Generation {generation}: Best Fitness = {fitness_function(best_solution)}")
    
    # Return the best solution after all generations
    best_solution = min(population, key=fitness_function)
    return best_solution

# Parameters
pop_size = 50
dim = 10
lower_bound = -5.0
upper_bound = 5.0
max_gen = 100
F = 0.8  # Mutation factor
Cr = 0.9  # Crossover rate

# Run the algorithm
best_solution = differential_evolution(pop_size, dim, lower_bound, upper_bound, max_gen, F, Cr)

print("Best solution found:", best_solution)
print("Fitness of the best solution:", fitness_function(best_solution))
````
