## Evolutionary Computation Tree
Evolutionary computation can be broadly classified into three major categories:

1. **Physic-inspired Evolutionary Computation**
   - Typical representative: Simulated annealing, inspired by the heating and cooling of materials.

2. **Darwin’s Evolution Theory Inspired Evolutionary Computation**
   - **Genetic Algorithm (GA)**: Inspired by natural selection, GA generates high-quality solutions to optimization problems through bio-inspired operators like mutation, crossover, and selection.
   - **Evolutionary Algorithms (EAs)**: Based on the survival of the fittest, these algorithms begin with a population of solutions that evolve over time using mechanisms such as genetic crossover and mutation.
   - **Evolutionary Strategy (ES)**: Developed by Schwefel in 1981, ES encodes DNA sequences using real numbers and adds mutation intensity for optimization.
   - **Evolutionary Programming (EP)**: Similar to ES but differs in mutation expressions and survival selections, with no crossover operator.
   - **Differential Evolution (DE)**: Developed by Storn and Price in 1997, DE uses mutation, crossover, and selection processes similar to GA but relies on the difference vector of the parent generation.

3. **Swarm Intelligence Inspired Evolutionary Computation**
   - **Ant Colony Optimization (ACO)**: Proposed by Dorigo in 1992, ACO utilizes the behavior of a swarm of ants to solve complex problems like finding the shortest path.
   - **Artificial Bee Colony (ABC) Algorithm**: Proposed by Karaboga in 2005, ABC is inspired by the foraging behavior of honey bees, where food source positions represent potential solutions.
   - **Particle Swarm Optimization (PSO)**: Developed by Kennedy and Eberhart in 1995, PSO is a stochastic algorithm that finds optimal regions of complex search spaces through population interactions.

## Multiobjective Optimization

In science and engineering applications, many optimization problems involve multiple objectives. Unlike single-objective problems, the objectives in multiobjective problems are usually in conflict. For example, in designing a complex hardware/software system, an optimal design might minimize cost and power consumption while maximizing overall performance. This structural contradiction necessitates different theories and methods for multiobjective optimization compared to single-objective optimization. The need for solving multiobjective optimization problems has led to the development of various evolutionary computation methods. This section discusses two types of multiobjective optimizations: multiobjective combinatorial optimization problems and multiobjective optimization problems.

## Multiobjective Combinatorial Optimization

Combinatorial optimization focuses on finding an optimal object from a finite set of objects. The general framework of multiobjective combinatorial optimization (MOCO) can be described as follows:

$$
\begin{align*}
\text{minimize} \quad & c^T_1 x = z_1(x), \ldots, c^T_K x = z_K(x), \\
\text{subject to} \quad & x \in D = \{x : x \in L_D, x \in B_n\}, \\
& L_D = \{x : Ax \leq b\},
\end{align*}
$$

where $ c_k \in \mathbb{R}^{n \times 1} $, the solution $ x = [x_1, \ldots, x_n]^T \in \mathbb{R}^n $ is a vector of discrete decision variables, $ A \in \mathbb{R}^{m \times n} $, $ b \in \mathbb{R}^m $, and $ B = \{0, 1\} $. The expression $ Ax \leq b $ denotes the elementwise inequality $ (Ax)_i \leq b_i $ for all $ i = 1, \ldots, m $.

Most multiobjective combinatorial optimization problems are related to:
- **Assignment Problems (APs)**
- **Transportation Problems (TPs)**
- **Network Flow Problems (NFPs)**

### 1. Multiobjective Assignment Problems

The formulation for multiobjective assignment problems is given by:

$$
\begin{align*}
\text{min} \quad & z_k(x) = \sum_{i=1}^n \sum_{j=1}^n c_{ij} x_{ij}, \quad k = 1, \ldots, K, \\
\text{subject to} \quad & \sum_{j=1}^n x_{ij} = 1, \quad i = 1, \ldots, n, \\
& \sum_{i=1}^n x_{ij} = 1, \quad j = 1, \ldots, n, \\
& x_{ij} \in \{0, 1\}, \quad i, j = 1, \ldots, n.
\end{align*}
$$

### 2. Multiobjective Transportation Problems

The formulation for multiobjective transportation problems is given by:

$$
\begin{align*}
\text{min} \quad & z_k(x) = \sum_{i=1}^r \sum_{j=1}^s c_{ij} x_{ij}, \quad k = 1, \ldots, K, \\
\text{subject to} \quad & \sum_{j=1}^s x_{ij} = a_i, \quad i = 1, \ldots, n, \\
& \sum_{i=1}^r x_{ij} = b_j, \quad j = 1, \ldots, n, \\
& x_{ij} \geq 0 \text{ and integer}, \quad i = 1, \ldots, r; \quad j = 1, \ldots, q.
\end{align*}
$$


## Multiobjective Network Flow Problems

Let $ G(N, A) $ be a network flow with a node set $ N $ and an arc set $ A $. The model can be stated as follows:

$$
\begin{align*}
\text{min} \quad & z_k(x) = \sum_{(i,j) \in A} c_{ij} x_{ij}, \quad k = 1, \ldots, K, \\
\text{subject to} \quad & \sum_{j \in A\setminus i} x_{ij} - \sum_{j \in A\setminus i} x_{ji} = 0, \quad \forall i \in N, \\
& l_{ij} \leq x_{ij} \leq u_{ij}, \quad \forall (i,j) \in A, \\
& x_{ij} \geq 0 \text{ and integer.}
\end{align*}
$$

Where:
- $ x_{ij} $ is the flow through arc $ (i, j) $,
- $ c_{ij}^k $ is the linear transhipment cost for arc $ (i, j) $ in objective $ k $,
- $ l_{ij} $ and $ u_{ij} $ are the lower and upper bounds on $ x_{ij} $, respectively.

## Multiobjective Min Sum Location Problems

The formulation for multiobjective min-sum location problems is given by:

$$
\begin{align*}
\text{min} \quad & \sum_{i=1}^m \sum_{k=1}^m f_{ik} z_i + \sum_{i=1}^m \sum_{j=1}^n c_{ij} x_{ij}, \quad k = 1, \ldots, K, \\
\text{subject to} \quad & \sum_{i=1}^m x_{ij} = 1, \quad j = 1, \ldots, n, \\
& \sum_{j=1}^n a_i z_i \leq d_{ij} x_{ij} \leq b_i z_i, \quad i = 1, \ldots, m, \\
& x_{ij} \in \{0, 1\}, \quad i = 1, \ldots, m; \quad j = 1, \ldots, n, \\
& z_i \in \{0, 1\}, \quad i = 1, \ldots, m.
\end{align*}
$$

Where:
- $ z_i = 1 $ if a facility is established at site $ i $,
- $ x_{ij} = 1 $ if customer $ j $ is assigned to the facility at site $ i $,
- $ d_{ij} $ is a certain usage of the facility at site $ i $ by customer $ j $,
- $ a_i $ and $ b_i $ are limitations on the total customer usage permitted at facility $ i $,
- $ c_{ij}^k $ is a variable cost (or distance, etc.) if customer $ j $ is assigned to facility $ i $,
- $ f_{ik} $ is a fixed cost associated with the facility at site $ i $.

## Challenges in Multiobjective Combinatorial Optimization

An unconstrained combinatorial optimization problem $ P = (S, f) $ is defined as an optimization problem where $ S $ is a finite set of solutions (called the search space), and $ f: S \to \mathbb{R}^+ $ is an objective function that assigns a positive cost value to each solution vector $ s \in S $. The goal is to find a solution of minimum cost value or a sufficiently good solution in a reasonable amount of time, especially in the case of approximate solution techniques.

The difficulty of multiobjective combinatorial optimization problems arises from two main factors:
1. Intensive cooperation is required with the decision maker to solve a multiobjective combinatorial optimization problem, leading to high requirements for effective tools used to generate efficient solutions.
2. Many combinatorial problems are hard even in single-objective versions; their multiobjective versions are often more complex.

A multiobjective decision-making (MODM) problem is ill-posed from a mathematical point of view because, except in trivial cases, it has no optimal solution. The goal of MODM methods is to find a solution that is most consistent with the decision maker’s preferences, i.e., the best compromise. Under very weak assumptions about the decision maker’s preferences, the best compromise solution belongs to the set of efficient solutions.


```python
import random
import matplotlib.pyplot as plt

# Parameters
num_individuals = 100   # Population size
num_generations = 50     # Number of generations
num_variables = 10       # Number of decision variables

# Objective functions
def objective1(individual):
    """Objective 1: Minimize the sum of squares."""
    return sum(x ** 2 for x in individual)

def objective2(individual):
    """Objective 2: Minimize the sum of (x-2)^2."""
    return sum((x - 2) ** 2 for x in individual)

# Generate a random individual
def generate_individual():
    return [random.uniform(0, 10) for _ in range(num_variables)]

# Create an initial population
population = [generate_individual() for _ in range(num_individuals)]

# Evaluate the population
def evaluate_population(population):
    return [(objective1(ind), objective2(ind)) for ind in population]

# Perform crossover
def crossover(parent1, parent2):
    crossover_point = random.randint(1, num_variables - 1)
    return parent1[:crossover_point] + parent2[crossover_point:]

# Perform mutation
def mutate(individual):
    mutation_index = random.randint(0, num_variables - 1)
    individual[mutation_index] = random.uniform(0, 10)
    return individual

# Main loop for the genetic algorithm
for generation in range(num_generations):
    # Evaluate fitness
    fitness = evaluate_population(population)
    
    # Selection (tournament selection)
    selected = []
    for _ in range(num_individuals):
        tournament = random.sample(population, 2)
        if fitness[population.index(tournament[0])][0] < fitness[population.index(tournament[1])][0]:
            selected.append(tournament[0])
        else:
            selected.append(tournament[1])
    
    # Crossover and mutation
    next_population = []
    for i in range(0, num_individuals, 2):
        parent1 = selected[i]
        parent2 = selected[i + 1]
        child1 = crossover(parent1, parent2)
        child2 = crossover(parent2, parent1)
        next_population.append(mutate(child1))
        next_population.append(mutate(child2))

    population = next_population

# Evaluate final population
final_fitness = evaluate_population(population)

# Plotting the results
objective1_values = [fit[0] for fit in final_fitness]
objective2_values = [fit[1] for fit in final_fitness]

plt.figure(figsize=(10, 6))
plt.scatter(objective1_values, objective2_values, c='blue', label='Solutions')
plt.title('Pareto Front for Multiobjective Optimization')
plt.xlabel('Objective 1: Minimize sum of squares')
plt.ylabel('Objective 2: Minimize sum of (x-2)^2')
plt.legend()
plt.grid()
plt.show()
```

##  Multiobjective Optimization Problems

Consider optimization problems with multiple objectives that may usually be incompatible.

**Definition .1 (Multiobjective Optimization Problem [78, 150])** A multiobjective optimization problem (MOP) consists of a set of $ n $ parameters (i.e., decision variables), a set of $ m $ objective functions, and a set of $ p $ inequality constraints and $ q $ equality constraints, and can be mathematically formulated as:

**Minimize/Maximize:**
$$
y = f(x) = [f_1(x), \ldots, f_m(x)]^T \tag{9.2.20}
$$

**Subject to:**
$$
g_i(x) \leq 0 \quad (i = 1, \ldots, p) \quad \text{or} \quad g(x) \leq 0 \tag{9.2.21}
$$
$$
h_j(x) = 0 \quad (j = 1, \ldots, q) \quad \text{or} \quad h(x) = 0 \tag{9.2.22}
$$



Here, the integer $ m \geq 2 $ is the number of objectives; $ x = [x_1, \ldots, x_n]^T \in X $ is a decision vector, $ y = [y_1, \ldots, y_m]^T $ is the objective vector; $ f_i : \mathbb{R}^n \to \mathbb{R} $, $ i = 1, \ldots, m $ are the objective functions, and $ g_i, h_j : \mathbb{R}^n \to \mathbb{R} $, $ i = 1, \ldots, m $; $ j = 1, \ldots, q $ are $ p $ inequality constraints and $ q $ equality constraints of the problem, respectively; while $ X = \{x \in \mathbb{R}^n\} $ is called the decision space, and $ Y = \{y \in \mathbb{R}^m | y = f(x), x \in X\} $ is known as the objective space.

The decision space and the objective space are two Euclidean spaces in MOPs:
- The $ n $-dimensional decision space $ X \in \mathbb{R}^n $ in which each coordinate axis corresponds to a component (called decision variable) of decision vector $ x $.
- The $ m $-dimensional objective space $ Y \in \mathbb{R}^m $ in which each coordinate axis corresponds to a component (i.e., a scalar objective function) of objective function vector $ y = f(x) = [f_1(x), \ldots, f_m(x)]^T $.

The MOP’s evaluation function $ f : x \to y $ maps decision vector $ x = [x_1, \ldots, x_n]^T $ to objective vector $ y = [y_1, \ldots, y_m]^T = [f_1(x), \ldots, f_m(x)]^T $. If some objective function $ f_i $ is to be maximized (or minimized), the original problem is written as the minimization (or maximization) of the negative objective $ -f_i $. Thereafter, we consider only the multiobjective minimization (or maximization) problems, rather than optimization problems mixing both minimization and maximization. The constraints $ g(x) \leq 0 $ and $ h(x) = 0 $ determine the set of feasible solutions.

**Definition.2 (Feasible Set)** The feasible set $ X_f $ is defined as the set of decision vectors $ x $ that satisfy the inequality constraint $ g(x) \leq 0 $ and the equality constraint $ h(x) = 0 $, i.e.,

$$
X_f = \{ x \in X | g(x) \leq 0; h(x) = 0 \}. \tag{9.2.23}
$$

### Example 1
The dynamic economic emission dispatch problem is a typical multiobjective optimization which attempts to minimize both cost function and emission function simultaneously while satisfying equality and inequality constraints [8].

1. **Objectives**:
   - **Cost function**: The fuel cost function of each thermal generator, considering the valve-point effect, is expressed as the sum of a quadratic and a sinusoidal function. The total fuel cost in terms of real power can be expressed as:

   $$
   f_1 = \sum_{m=1}^{M} \sum_{i=1}^{N} \left( a_i + b_i P_{im} + c_i P_{im}^2 + d_i \sin(e_i (P_{imin} - P_{im})) \right), \tag{9.2.24}
   $$

   where $ N $ is the number of generating units, $ M $ is the number of hours in the time horizon; $ a_i, b_i, c_i, d_i, e_i $ are cost coefficients of the $ i $-th unit; $ P_{im} $ is the power output of the $ i $-th unit at time $ m $, and $ P_{imin} $ is the lower generation limits for the $ i $-th unit.

   - **Emission function**: The atmospheric pollutants such as sulfur oxides (SOx) and nitrogen oxides (NOx) caused by fossil-fueled generating units can be modeled separately. However, for comparison purposes, the total emission of these pollutants which is the sum of a quadratic and an exponential function can be expressed as:

   $$
   f_2 = \sum_{m=1}^{M} \sum_{i=1}^{N} \left( \alpha_i + \beta_i P_{im} + \gamma_i P_{im}^2 + \eta_i \exp(\delta_i P_{im}) \right), \tag{9.2.25}
   $$

   where $ \alpha_i, \beta_i, \gamma_i, \eta_i, \delta_i $ are emission coefficients of the $ i $-th unit.

2. **Constraints**:
   - **Real power balance constraints**: The total real power generation must balance the predicted power demand plus the real power losses in the transmission lines, at each time interval over the scheduling horizon, i.e.,

   $$
   \sum_{i=1}^{N} P_{im} - PD_m - PL_m = 0, \quad m = 1, \ldots, M, \tag{9.2.26}
   $$

   where $ PD_m $ is the load demand at time $ m $, $ PL_m $ is the transmission line losses at time $ m $.

   - **Real power operating limits**: 

   $$
   P_{imin} \leq P_{im} \leq P_{imax}, \quad i = 1, \ldots, N, \quad m = 1, \ldots, M, \tag{9.2.27}
   $$

   where $P_{imax} $ is the upper generation limits for the $ i $-th unit.

   - **Generating unit ramp rate limits**:

   $$
   P_{im} - P_{i(m-1)} \leq UR_i, \quad i = 1, \ldots, N, \quad m = 1, \ldots, M, \tag{9.2.28}
   $$

   $$
   P_{i(m-1)} - P_{im} \leq DR_i, \quad i = 1, \ldots, N, \quad m = 1, \ldots, M, \tag{9.2.29}
   $$

   where $ UR_i $ and $ DR_i $ are ramp-up and ramp-down rate limits of the $ i $-th unit, respectively.


```python
import random
from deap import base, creator, tools, algorithms

# Define the problem as a minimization problem with two objectives
creator.create("FitnessMin", base.Fitness, weights=(-1.0, -1.0))  # Minimize both objectives
creator.create("Individual", list, fitness=creator.FitnessMin)

# Define the evaluation function
def evaluate(individual):
    x = individual[0]
    y = individual[1]
    f1 = x**2 + y**2  # First objective (minimize distance to origin)
    f2 = (x - 1)**2 + y**2  # Second objective (minimize distance to point (1, 0))
    return f1, f2

# Create the toolbox
toolbox = base.Toolbox()
toolbox.register("individual", tools.initRepeat, creator.Individual, lambda: random.uniform(-5, 5), 2)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
toolbox.register("select", tools.selNSGA2)

# Set parameters
population_size = 100
generations = 50
crossover_prob = 0.7
mutation_prob = 0.2

# Create an initial population
population = toolbox.population(n=population_size)

# Run the optimization
for gen in range(generations):
    # Select the next generation individuals
    offspring = toolbox.select(population, len(population))
    offspring = list(map(toolbox.clone, offspring))

    # Apply crossover and mutation
    for child1, child2 in zip(offspring[::2], offspring[1::2]):
        if random.random() < crossover_prob:
            toolbox.mate(child1, child2)
            del child1.fitness.values
            del child2.fitness.values

    for mutant in offspring:
        if random.random() < mutation_prob:
            toolbox.mutate(mutant)
            del mutant.fitness.values

    # Evaluate the individuals with an invalid fitness
    invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
    fitnesses = map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit

    # Replace the old population by the offspring
    population[:] = offspring

# Print the results
front = tools.sortNondominated(population, len(population), first_front_only=True)[0]
for individual in front:
    print(f"Individual: {individual}, Objectives: {individual.fitness.values}")
```

## Multiobjective Optimization Problems

### Definition .3: Ideal Vector
Let $ \mathbf{x}_{\text{opt},k} = [x_1, \ldots, x_n]^T $ be a vector of variables which optimizes (either minimizes or maximizes) the $ k $th objective function $ f_k(\mathbf{x}) $. In other words, if the vector $ \mathbf{x}_{\text{opt},k} \in \mathcal{X}_f $ is such that 

$$
\text{opt}_{k} f_k = \text{opt} f_k(\mathbf{x}), \quad k \in \{1, \ldots, m\}, \tag{9.2.30}
$$

then the objective vector $ \mathbf{f}_{\text{opt}} = [f_1, \ldots, f_m]^T $ (where $ f_k $ denotes the optimum of the $ k $th objective function) is ideal for a multiobjective optimization problem (MOP). The point in $ \mathbb{R}^n $ that determines this vector is the ideal solution, which is consequently called the ideal vector.

However, due to objective conflict and/or interdependence among the $ n $ decision variables of $ \mathbf{x} $, such an ideal solution vector is often unattainable for MOPs: none of the feasible solutions allows simultaneous optimal solutions for all objectives. In other words, individual optimal solutions for each objective are usually different. Thus, a mathematically most favorable solution should offer the least objective conflict.

### Examples of Multiobjective Optimization Problems

#### Traveling Salesman Problem (TSP)
The Traveling Salesman Problem can be stated as:

$$
\min_{i \in \rho(i)} c_i^{(k)}, \quad k = 1, \ldots, K, \tag{9.2.31}
$$

where $ \rho(i) $ is a tour of $ \{1, \ldots, n\} $, and $ k $ denotes trip $ k $. Let $ C = \{c_1, \ldots, c_{N_c}\} $ be a set of cities, where $ c_i $ is the $ i $th city and \( N_c \) is the number of all cities. The goal of TSP is to find the shortest length closed tour visiting each city only once.

##### Different Forms of TSP
1. **Euclidean TSP**: Cities $ r \in C $ are given by their coordinates $(x_r, y_r)$, and $ d(r, s) $ is the Euclidean distance between cities $ r $ and $ s $.
2. **Symmetric TSP**: If $ d(r, s) = d(s, r) $ for all $ (r, s) $, then the TSP becomes a symmetric TSP (STSP).
3. **Asymmetric TSP**: If $ d(r, s) \neq d(s, r) $ for at least some $ (r, s) $, then the TSP is an asymmetric TSP (ATSP).
4. **Dynamic TSP**: In a dynamic TSP (DTSP), cities can be added or removed at runtime.

#### Multiobjective Data Mining
Data mining tasks can be broadly classified into two categories:

1. **Predictive Techniques**: These learn from current data to make predictions about new data sets.
2. **Descriptive Techniques**: These provide a summary of the data.

Some commonly used data mining tasks include:

1. **Feature Selection**: Determines the optimum relevant set of features for classification or clustering. Formally defined as:

$$
P(F^*) = \min_{F \in \mathcal{F}} P(F, \mathcal{X}), \tag{9.2.32}
$$

where $ \mathcal{F} $ is the set of possible feature subsets.

2. **Classification**: Produces a function $ h: \mathcal{X} \to \mathcal{Y} $ that approximates a mapping $ g $ as closely as possible.

3. **Clustering**: Groups patterns into clusters such that patterns in the same cluster are similar and patterns in different clusters are dissimilar. 

4. **Association Rule Mining**: Discovers rules showing attribute-value associations in transaction data. The objective is to find all rules of the form 

$$
X \Rightarrow Y, \quad X \cap Y = \emptyset 
$$ 

with probability $ c\% $.

Most data mining problems can be thought of as optimization problems, often involving multiple criteria to optimize.

## Pareto Optimization Theory

Many real-world problems in artificial intelligence involve simultaneous optimization of several incommensurable and often competing objectives. In these cases, there may exist solutions in which the performance on one objective cannot be improved without reducing performance on at least one other. In other words, there is often no single optimal solution with respect to all objectives. In practice, there could be a number of optimal solutions in multiobjective optimization problems (MOPs) and the suitability of one solution depends on a number of factors including designer’s choice and problem environment. 

These solutions are optimal in the wider sense that no other solutions in the search space are superior to them when all objectives are considered. In these problems, the Pareto optimization approach is a natural choice. Solutions given by the Pareto optimization approach are called Pareto-optimal solutions that are closely related to Pareto concepts.

## Pareto Concepts

Pareto concepts are named after Vilfredo Pareto (1848–1923). These concepts constitute the Pareto optimization theory for multiple objectives. Consider the multiobjective optimization problems

$$
\max/\min_{x \in S} f(x) = [f_1(x), \ldots, f_m(x)]^T
$$

in which the objectives (or criterion functions) $f_1, \ldots, f_m$ are real-valued, continuous, and quasi-concave (for maximization) or quasi-convex (for minimization) on a closed convex set $S$. 

Three approaches can be taken to find the solution(s) to the above problems:

- **Reformulation approach**: This approach entails reformulating the problem as a single objective problem. To do so, additional information is required from the decision makers, such as the relative importance or weights of the objectives, goal levels for the objectives, value functions, etc.

- **Decision making approach**: This approach requires that the decision makers interact with the optimization procedure typically by specifying preferences between pairs of presented solutions.

- **Pareto optimization approach**: It finds a representative set of nondominated solutions approximating the Pareto front. Pareto optimization methods, such as evolutionary multiobjective optimization algorithms, allow decision makers to investigate the potential solutions without a priori judgments regarding the relative importance of objective functions. 

Post-Pareto analysis is necessary to select a single solution for implementation. On the other hand, the most distinguishing feature of single objective evolutionary algorithms (EAs) compared to other heuristics is that EAs work with a population of solutions, and thus are able to search for a set of solutions in a single run. Due to this feature, single objective evolutionary algorithms are easily extended to multiobjective optimization problems. Multiobjective evolutionary algorithms (MOEAs) have become one of the most active research areas in evolutionary computation. 

To describe the Pareto optimality of MOP solutions in which we are interested, it needs the following key Pareto concepts: Pareto dominance, Pareto optimality, the Pareto-optimal set, and the Pareto front. For the convenience of narration, denote 

$$
x = [x_1, \ldots, x_n]^T \in F \subseteq S 
$$ 

and 

$$
\bar{x} = [\bar{x}_1, \ldots, \bar{x}_n]^T \in F \subseteq S,
$$ 

where $F$ is the feasible region, in which the constraints are satisfied. 

### Definition.4 (Vector Function Relation)

For any two objective vectors \(f(x)\) and \(f(\bar{x})\), their relations are written as 

$$
f(x) = f(\bar{x}) \iff f_i(x) = f_i(\bar{x}), \; \forall i = 1, \ldots, m; 
$$

$$
f(x) \neq f(\bar{x}) \iff f_i(x) \neq f_i(\bar{x}), \; \text{for at least one } i \in \{1, \ldots, m\};
$$ 

$$
f(x) \leq f(\bar{x}) \iff f_i(x) \leq f_i(\bar{x}), \; \forall i = 1, \ldots, m; 
$$ 

$$
f(x) < f(\bar{x}) \iff f(x) \leq f(\bar{x}) \land f(x) \neq f(\bar{x});
$$ 

$$
f(x) \geq f(\bar{x}) \iff f_i(x) \geq f_i(\bar{x}), \; \forall i = 1, \ldots, m; 
$$ 

$$
f(x) > f(\bar{x}) \iff f(x) \geq f(\bar{x}) \land f(x) \neq f(\bar{x}).
$$ 

Here the symbol $\land$ may be interpreted as a logical conjunction operator (“AND”). It goes without saying that the above relations are available for $f(x) = x$ and $f(\bar{x}) = \bar{x}$ as well. However, for two decision vectors $x$ and $\bar{x}$, these relations are meaningless.

### Definition.5 (Pareto Dominance)

For any two decision vectors $x$ and $\bar{x}$ in minimization problems, $x$ is said to dominate (or outperform) $\bar{x}$, denoted by $x \prec \bar{x}$, if and only if their objectives or outcomes satisfy jointly the following two conditions:

1. $x$ is no worse than $\bar{x}$ in all objective functions, i.e., $f_i(x) \leq f_i(\bar{x})$ for all $i \in \{1, \ldots, m\},$
2. $x$ is strictly better than $\bar{x}$ for at least one objective function, i.e., $f_j(x) < f_j(\bar{x})$ for at least one $j \in \{1, \ldots, m\}.$

or is equivalently represented as 

$$
x \prec \bar{x} \iff \forall i \in \{1, \ldots, m\} : f_i(x) \leq f_i(\bar{x}) \land \exists j \in \{1, \ldots, m\} : f_j(x) < f_j(\bar{x}), \tag{9.3.8}
$$ 

or simply written as 

$$
x \prec \bar{x} \iff f(x) < f(\bar{x}). \tag{9.3.9}
$$ 

Similarly, for maximization problems, 

$$
x \prec \bar{x} \iff \forall i \in \{1, \ldots, m\} : f_i(x) \geq f_i(\bar{x}) \land \exists j \in \{1, \ldots, m\} : f_j(x) > f_j(\bar{x}), \tag{9.3.10}
$$ 

or simply written as 

$$
x \prec \bar{x} \iff f(x) > f(\bar{x}). \tag{9.3.11}
$$ 

The Pareto dominance can be either weak or strong.

### Definition .6 (Weak Pareto Dominance)

A decision vector $x$ is said to weakly dominate another decision vector $\bar{x}$, denoted by $x \preceq \bar{x}$, if and only if 

$$
x \preceq \bar{x} \iff f(x) \leq f(\bar{x}) \quad (\text{for minimization}), \tag{9.3.12}
$$ 

$$
x \preceq \bar{x} \iff f(x) \geq f(\bar{x}) \quad (\text{for maximization}). \tag{9.3.13}
$$ 

### Definition.7 (Strict Pareto Dominance)

A decision vector $x$ is said to strictly dominate $\bar{x}$, denoted by $x \prec \bar{x}$, if and only if $f_i(x)$ is better than $f_i(\bar{x})$ in all objectives, i.e., 

$$
x \prec \bar{x} \iff \forall i \in \{1, \ldots, m\} : f_i(x) < f_i(\bar{x}) \quad (\text{for minimization}), \tag{9.3.14}
$$ 

$$
x \prec \bar{x} \iff \forall i \in \{1, \ldots, m\} : f_i(x) > f_i(\bar{x}) \quad (\text{for maximization}). \tag{9.3.15}
$$ 

### Definition.8 (Incomparable)

Two decision vectors $x$ and $\bar{x}$ are said to be incomparable, denoted by $x \sim \bar{x}$, if neither $x$ weakly dominates $\bar{x}$ nor $\bar{x}$ weakly dominates $x$, i.e., 

$$
x \sim \bar{x} \iff x \npreceq \bar{x} \land \bar{x} \npreceq x. \tag{9.3.16}
$$ 

Two incomparable solutions $x$ and $\bar{x}$ are also known as mutually nondominating.

### Definition.9 (Nondominance)

A decision vector $x \in X_f$ is said to be nondominated regarding a set $A \subseteq X_f$ if and only if there is no vector $a$ in $A$ which dominates $x$; formally 

$$
\forall a \in A : a \nprec x. \tag{9.3.17}
$$ 

In other words, a solution $x$ is said to be a nondominated solution if there is no $a$ in the subset $A$ that dominates $x$. 

A Pareto improvement is a change to a different allocation that makes at least one objective better off without making any other objective worse off. A trial solution is known as “Pareto efficient” or “Pareto-optimal” or “globally nondominated” when no further Pareto improvements can be made, in which case we are assumed to have reached Pareto optimality. 

### Definition 9.10 (Pareto-Optimality)

A decision vector $x \in X_f$ is said to be Pareto-optimal if and only if $x$ is nondominated regarding the feature region $X_f$, namely 

$$
\forall a \in X_f : a \nprec x. \tag{9.3.18}
$$ 

In other words, all decision vectors which are not dominated by any other decision vector are known as “nondominated” or Pareto optimal. The phrase “Pareto optimal” means the solution is optimal with respect to the entire decision variable space \(F\) unless otherwise specified. 

Fig.2, by inferencing Economics Wiki on “Pareto Efficient and Pareto Optimal Tutorial,” shows the relation of the Pareto efficiency and Pareto improvement for the two-objective minimization problems 

$$
\min_{x} f(x) = [f_1(x), f_2(x)]^T. 
$$ 

A nondominated decision vector in $A \subseteq X_f$ is only Pareto-optimal in a local decision space $A$, while a Pareto-optimal decision vector is Pareto-optimal in the entire feature decision space $X_f$. Therefore, a Pareto-optimal decision vector is definitely nondominated, but a nondominated decision vector is not necessarily Pareto-optimal. 

However, to find Pareto-optimal solutions still is not the ultimate aim of multiobjective optimization. This is because:

- There may be a large number of Pareto-optimal solutions. To gain deeper insight into the multiobjective optimization problem and knowledge about alternate solutions, there is often a special interest in finding or approximating the Pareto-optimal set.
- Due to conflicting objectives, it is only possible to obtain a set of trade-off solutions (referred to as the Pareto-optimal set in decision space or Pareto-optimal front in objective space), instead of a single optimal solution. 

Therefore, our aim is to find the decision vectors that are nondominated within the entire search space. These decision vectors constitute the so-called Pareto-optimal set, as defined below.




Fig.2 Pareto efﬁciency and Pareto improvement. Point A is an inefﬁcient allocation between preference criterions f1 and f2 because it does not satisfy the constraint curve of f1 and f2 . Two decisions to move from Point A to Points C and D would be a Pareto improvement, respectively. They improve both f1 and f2 , without making anyone else worse off. Hence, these two moves would be a Pareto improvement and be Pareto optimal, respectively. A move from Point A to Point B would not be a Pareto improvement because it decreases the cost f1 by increasing another cost f2 , thus making one side better off by making another worse off. The move from any point that lies under the curve to any point on the curve cannot be a Pareto improvement due to making one of two criterions f1 and f2 worse


```python
# Required imports for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Sample function to evaluate objectives
def objectives(x):
    # Objective 1: f1(x) = x
    f1 = x
    # Objective 2: f2(x) = (1 - x)^2
    f2 = (1 - x) ** 2
    return f1, f2

# Generate a set of candidate solutions
def generate_solutions(num_solutions=100):
    return [i / num_solutions for i in range(num_solutions + 1)]

# Determine Pareto optimal solutions
def pareto_optimal(solutions):
    pareto_front = []
    for s1 in solutions:
        is_dominated = False
        for s2 in solutions:
            if s1 != s2:
                f1_s1, f2_s1 = objectives(s1)
                f1_s2, f2_s2 = objectives(s2)

                # Check if s2 dominates s1
                if f1_s2 <= f1_s1 and f2_s2 <= f2_s1:
                    is_dominated = True
                    break
        if not is_dominated:
            pareto_front.append(s1)
    return pareto_front

# Visualize Pareto front
def plot_pareto_front(pareto_front):
    # Calculate objectives for Pareto optimal solutions
    f1_values = [objectives(s)[0] for s in pareto_front]
    f2_values = [objectives(s)[1] for s in pareto_front]

    plt.figure(figsize=(10, 6))
    plt.scatter(f1_values, f2_values, color='b', label='Pareto-optimal Solutions')
    plt.title('Pareto Front')
    plt.xlabel('Objective 1: f1(x)')
    plt.ylabel('Objective 2: f2(x)')
    plt.axhline(0, color='grey', lw=0.5, ls='--')
    plt.axvline(0, color='grey', lw=0.5, ls='--')
    plt.legend()
    plt.grid()
    plt.show()

# Main execution
if __name__ == "__main__":
    num_solutions = 100
    solutions = generate_solutions(num_solutions)
    pareto_front = pareto_optimal(solutions)
    plot_pareto_front(pareto_front)
```

## Multi-Objective Optimization Concepts

## Definitions

**Definition.11 (Nondominated Sets [150])**  
Let $ A \subseteq X_f $ and the function $ g(A) $ give the set of nondominated decision vectors in $ A $:
$$
g(A) = \{ a \in A \ | \ a \text{ is nondominated regarding } A \} \tag{9.3.19}
$$
The set $ g(A) $ is said to be the **nondominated set (NS)** with respect to $ A $. When $ A = X_f $, the corresponding set:
$$
\begin{align*}
N S_{\text{min}} &= \{ x \in X_f \ | \ \{ f(x') \in Y \ | \ f(x') \leq f(x) \land f(x') \neq f(x) \} = \emptyset \} \tag{9.3.20} \\
N S_{\text{max}} &= \{ x \in X_f \ | \ \{ f(x') \in Y \ | \ f(x') \geq f(x) \land f(x') \neq f(x) \} = \emptyset \} \tag{9.3.21}
\end{align*}
$$
is known as the nondominated set in the entire decision space for minimization and maximization, respectively.

**Definition.12 (Pareto-Optimal Set [150])**  
The **Pareto-optimal set (PS)** is the set of all objectives given by nondominated decision vectors \( x \) in the nondominated set \( NS \):
$$
\begin{align*}
P S_{\text{min}} &= \{ f(x) \in Y \ | \ \{ f(x') \in Y \ | \ f(x') \leq f(x) \land f(x') \neq f(x) \} = \emptyset \} \tag{9.3.22} \\
P S_{\text{max}} &= \{ f(x) \in Y \ | \ \{ f(x') \in Y \ | \ f(x') \geq f(x) \land f(x') \neq f(x) \} = \emptyset \} \tag{9.3.23}
\end{align*}
$$
Pareto-optimal decision vectors cannot be improved in any objective without causing a degradation in at least one other objective; they represent globally optimal solutions.

**Definition 9.13 (Local Pareto-Optimal Set [23, 171])**  
Consider a set of decision vectors $ X' \subseteq X_f $. The set $ X' $ is said to be a local Pareto-optimal set if and only if 
$$
\forall x' \in X', \ \exists x \in X_f \ : \ x \prec x' \land \| x - x' \| < \epsilon \land \forall i \in \{1, \ldots, m\} : |f_i(x) - f_i(x')| < \delta, \tag{9.3.24}
$$
where $ \| \cdot \| $ is a corresponding distance metric and $ \epsilon > 0, \delta > 0 $.

**Definition.14 (Global Pareto-Optimal Set [23, 171])**  
The set $ X' $ is called a global Pareto-optimal set if and only if 
$$
\forall x' \in X' : \exists x \in X_f : x \prec x'. \tag{9.3.25}
$$

The set of all objectives given by the Pareto-optimal set of decision vectors is known as the **Pareto-optimal front (or simply Pareto front)**.

**Definition 9.15 (Pareto Front)**  
Given an objective vector $ f(x) = [f_1(x), \ldots, f_m(x)]^T $ and a Pareto-optimal set \( PS \), the Pareto front \( PF \) is defined as:
$$
PF = \{ f(x) \in \mathbb{R}^m \ | \ x \in PS \}. \tag{9.3.26}
$$
Solutions in the Pareto front represent the possible optimal trade-offs between competing objectives.

**Definition 9.16 (Nadir Objective Vector, Ideal Objective Vector)**  
The **nadir objective vector** is defined as:
$$
z_{\text{nad}} = \sup f_i(x), \quad \forall i = 1, \ldots, K \quad (x \text{ is Pareto optimal}) \tag{9.3.27}
$$
and the **ideal objective vector** as:
$$
z_{\text{ideal}} = \inf f_i(x), \quad \forall i = 1, \ldots, K. \tag{9.3.28}
$$

In other words, the components of a nadir and an ideal objective vector define upper and lower bounds for the objective function values of Pareto-optimal solutions, respectively.

**Definition 9.17 (Archive)**  
The nondominated set produced by heuristic algorithms is referred to as the **archive** (denoted by \( A \)) of the estimated Pareto front. The archive of estimated Pareto fronts will only be an approximation to the true Pareto front.

### Selection Methods for Pareto-Optimal Solutions
When solving multi-objective optimization problems, since there are many possible Pareto-optimal solutions rather than a single optimal solution for possibly contradictory multi-objective functions, we need to organize or partition found solutions for selecting appropriate solution(s). The selection methods for Pareto-optimal solutions are divided into the following four categories:
- Fitness selection approach
- Nondominated sorting approach
- Crowding distance assignment approach
- Hierarchical clustering approach


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to generate random decision vectors
def generate_decision_vectors(num_vectors, dimension):
    return np.random.rand(num_vectors, dimension)

# Objective functions
def objective_function_1(x):
    return x[:, 0] ** 2 + x[:, 1] ** 2  # Example: Minimize x1^2 + x2^2

def objective_function_2(x):
    return (x[:, 0] - 1) ** 2 + (x[:, 1] - 1) ** 2  # Example: Minimize (x1-1)^2 + (x2-1)^2

# Function to find nondominated solutions
def get_nondominated_solutions(objectives):
    is_nondominated = np.ones(objectives.shape[0], dtype=bool)
    for i in range(objectives.shape[0]):
        for j in range(objectives.shape[0]):
            if (objectives[j, 0] <= objectives[i, 0] and objectives[j, 1] < objectives[i, 1]) or \
               (objectives[j, 0] < objectives[i, 0] and objectives[j, 1] <= objectives[i, 1]):
                is_nondominated[i] = False
                break
    return is_nondominated

# Generate decision vectors
num_vectors = 100
dimension = 2
decision_vectors = generate_decision_vectors(num_vectors, dimension)

# Evaluate objectives
objectives = np.array([objective_function_1(decision_vectors), objective_function_2(decision_vectors)]).T

# Find nondominated solutions
nondominated_mask = get_nondominated_solutions(objectives)
nondominated_solutions = objectives[nondominated_mask]

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=objectives[:, 0], y=objectives[:, 1], label='All Solutions', alpha=0.5)
sns.scatterplot(x=nondominated_solutions[:, 0], y=nondominated_solutions[:, 1], color='red', label='Nondominated Solutions', s=100)
plt.title('Pareto Front')
plt.xlabel('Objective 1')
plt.ylabel('Objective 2')
plt.legend()
plt.grid()
plt.show()
```

## Fitness Selection Approach

To fairly evaluate the quality of each solution and direct the solution search towards the Pareto-optimal solution, fitness assignment plays a key role. Various benchmark metrics, also known as indicators, are commonly used to evaluate the quality of Pareto fronts. These metrics are:

1. **Hypervolume**
2. **Spacing**
3. **Maximum Spread**
4. **Coverage**

## Metrics and Indicators for Fitness Selection

### Hypervolume (HV)
The hypervolume for a nondominated vector $ x_i \in P_a^* $ is defined as:

$$
\text{HV} = \sum_{i} a_i \; \Big| \; x_i \in P_a^*,
$$

where $ P_a^* $ is the area in the objective search space covered by the obtained Pareto front, and $ a_i $ is the hypervolume determined by the components of $ x_i $ and the origin.

### Spacing (S)

Spacing is a measure of the spread or distribution of nondominated solutions across the Pareto front. It measures the variance in the distance between adjacent nondominated solutions:

$$
S = \sqrt{\frac{1}{m-1} \sum_{i=1}^m (d - d_i)^2},
$$

where:
- $ d $ is the mean distance between adjacent solutions,
- $ m $ is the number of nondominated solutions,
- $ d_i = \min_{j} \left( |f_{1i}(x) - f_1(x)| + |f_{2i}(x) - f_2(x)| \right), \; i, j = 1, \ldots, m $.

A spacing value of zero indicates that all solutions are equidistantly spaced in the Pareto front.

### Maximum Spread (MS)

This metric evaluates the maximum extension covered by the nondominated solutions in the Pareto front:

$$
\text{MS} = \sum_{i=1}^K \left( \max \left( f_{i1}, \ldots, f_{im} \right) - \min \left( f_{i1}, \ldots, f_{im} \right) \right),
$$

where $ m $ is the number of solutions in the Pareto front, and $ K $ is the number of objectives in the problem.

Higher values of MS indicate better performance in terms of diversity.

### Coverage (C)

The coverage metric, proposed by Zitzler and Thiele, is denoted by $ C(A, B) $ and maps the ordered pair $ (A, B) $ to the interval $ [0, 1] $:

$$
C(A, B) = \frac{| \{ b \in B \; | \; \exists a \in A : a \succeq b \} |}{|B|},
$$

where $ |B| $ is the number of elements in the set $ B $. A value of $ C(A, B) = 1 $ implies that all solutions in $ B $ are weakly dominated by solutions in $ A $, while $ C(A, B) = 0 $ implies that none of the solutions in $ B $ is weakly dominated by $ A $.

**Note**: Both $ C(A, B) $ and $ C(B, A) $ should be evaluated, as $ C(A, B) $ is not necessarily equal to $ 1 - C(B, A) $. If $ 0 < C(A, B) < 1 $ and $ 0 < C(B, A) < 1 $, then the sets $ A $ and $ B $ are incomparable.

## Fitness Selection Techniques

Two popular fitness selection approaches are binary tournament selection and roulette wheel selection.

### 1. Binary Tournament Selection

In binary tournament selection, $ k $ individuals are drawn from the population to compete, and the best individual among them is selected. The number $ k $ is the tournament size, often set to 2. This selection strategy is typically known as binary tournament selection.

### 2. Roulette Wheel Selection

Roulette wheel selection, also known as fitness proportionate selection, is a method that assigns a probability of selection to each individual based on its fitness. Given the fitness $ F_i $ of individual $ i $ in the population, the probability $ p_i $ of selecting $ i $ is:

$$
p_i = \frac{F_i}{\sum_{j=1}^N F_j},
$$

where $N $ is the number of individuals in the population.

#### Algorithm 9.1: Roulette Wheel Selection Algorithm

The roulette wheel selection algorithm assigns probabilities based on fitness scores, then selects individuals according to these probabilities for recombination.


```python
import random

# Sample objectives for demonstration
# Assuming we have solutions with two objectives each for simplicity
solutions = [
    [0.5, 0.7], [0.2, 0.8], [0.8, 0.3],
    [0.6, 0.5], [0.4, 0.6], [0.7, 0.4]
]

# Hypervolume Calculation
def hypervolume(solutions):
    hv = 0
    origin = [0, 0]
    for sol in solutions:
        hv += (sol[0] - origin[0]) * (sol[1] - origin[1])
    return hv

# Spacing Calculation
def spacing(solutions):
    m = len(solutions)
    distances = []
    for i, sol1 in enumerate(solutions):
        min_distance = float('inf')
        for j, sol2 in enumerate(solutions):
            if i != j:
                distance = sum(abs(sol1[k] - sol2[k]) for k in range(len(sol1)))
                min_distance = min(min_distance, distance)
        distances.append(min_distance)
    d_bar = sum(distances) / m
    return (sum((d_bar - d) ** 2 for d in distances) / (m - 1)) ** 0.5

# Maximum Spread Calculation
def maximum_spread(solutions):
    min_values = [min(sol[i] for sol in solutions) for i in range(len(solutions[0]))]
    max_values = [max(sol[i] for sol in solutions) for i in range(len(solutions[0]))]
    return sum(max_values[i] - min_values[i] for i in range(len(min_values)))

# Coverage Calculation
def coverage(A, B):
    covered = sum(any(all(a[i] >= b[i] for i in range(len(b))) for a in A) for b in B)
    return covered / len(B)

# Binary Tournament Selection
def binary_tournament_selection(solutions):
    selected = []
    for _ in range(len(solutions) // 2):
        sol1, sol2 = random.sample(solutions, 2)
        if hypervolume([sol1]) >= hypervolume([sol2]):
            selected.append(sol1)
        else:
            selected.append(sol2)
    return selected

# Roulette Wheel Selection
def roulette_wheel_selection(solutions):
    fitness_values = [hypervolume([sol]) for sol in solutions]
    total_fitness = sum(fitness_values)
    selection_probs = [f / total_fitness for f in fitness_values]
    selected = []
    for _ in range(len(solutions)):
        r = random.random()
        cumulative_prob = 0
        for i, sol in enumerate(solutions):
            cumulative_prob += selection_probs[i]
            if r <= cumulative_prob:
                selected.append(sol)
                break
    return selected

# Run Metrics
print("Hypervolume:", hypervolume(solutions))
print("Spacing:", spacing(solutions))
print("Maximum Spread:", maximum_spread(solutions))
print("Coverage:", coverage(solutions, solutions))  # Self-coverage example

# Run Selection Methods
print("Binary Tournament Selection:", binary_tournament_selection(solutions))
print("Roulette Wheel Selection:", roulette_wheel_selection(solutions))
```

## Algorithm .1: Roulette Wheel Fitness Selection

### Step-by-Step Procedure

1. **Input**: Define the population size $ k $.

2. **Initialization**: Generate the initial population of candidate solutions randomly.

3. **Evaluate Fitness**:
   - Calculate the fitness $ F_i $ of each individual $ i $ in the population.

4. **Compute Selection Probability**:
   - Determine the probability (slot size) $ p_i $ for each member of the population:
     $$
     p_i = \frac{F_i}{\sum_{j=1}^{k} F_j}
     $$
   where $ k $ is the population size.

5. **Cumulative Probability**:
   - Calculate the cumulative probability $ q_i $ for each individual:
     $$
     q_i = \sum_{j=1}^{i} p_j
     $$

6. **Random Selection**:
   - Generate a uniform random number $ r \in (0, 1] $.
   - If $ r < q_1 $, then select the first chromosome $ x_1 $. Otherwise, select the individual $ x_i $ such that $ q_{i-1} < r \leq q_i $.

7. **Repeat**:
   - Repeat steps 4–7, $ k $ times to generate $ k $ candidate solutions in the mating pool.

8. **Output**: The mating pool contains $ k $ candidate solutions.

### Summary

This selection process creates a probability-based method for selecting candidates, with higher-fitness individuals having a greater chance of selection while maintaining diversity in the population.


```python
import random

def generate_initial_population(size):
    """Generate an initial population of random candidate solutions."""
    return [random.uniform(0, 1) for _ in range(size)]

def evaluate_fitness(population):
    """Evaluate the fitness of each individual in the population."""
    # In this example, fitness is simply the value of the individual.
    return [individual for individual in population]

def roulette_wheel_selection(population, fitness):
    """Perform roulette wheel selection to create a mating pool."""
    k = len(population)
    
    # Compute probabilities of selection
    total_fitness = sum(fitness)
    probabilities = [f / total_fitness for f in fitness]
    
    # Compute cumulative probabilities
    cumulative_probabilities = [sum(probabilities[:i + 1]) for i in range(k)]
    
    mating_pool = []
    
    # Generate candidates in the mating pool
    for _ in range(k):
        r = random.uniform(0, 1)
        for i, cp in enumerate(cumulative_probabilities):
            if r <= cp:
                mating_pool.append(population[i])
                break
    
    return mating_pool

# Main execution
population_size = 10  # Size of the population
initial_population = generate_initial_population(population_size)
fitness = evaluate_fitness(initial_population)
mating_pool = roulette_wheel_selection(initial_population, fitness)

print("Initial Population:", initial_population)
print("Fitness Values:", fitness)
print("Mating Pool:", mating_pool)
```

## Nondominated Sorting Approach

The nondominated selection is based on nondominated rank: dominated individuals are penalized by subtracting a small fixed penalty term from their expected number of copies during selection. However, this algorithm fails when the population has very few nondominated individuals, which may result in a large fitness value for those few nondominated points, ultimately leading to high selection pressure.

The idea behind the nondominated sorting procedure is twofold:
- Use a ranking selection method to emphasize good points.
- Use a niche method to maintain stable subpopulations of good points.

An individual (or solution) is assigned to the first nondominated level (or rank) if it is not dominated by all other individuals. An individual is said to have the second nondominated level or rank if it is dominated only by the individual(s) in the first nondominated level. Similarly, one can define individuals in the third or higher nondominated level or rank. It is noted that more individuals may have the same nondominated level or rank. Let $ \text{irank} $ represent the nondominated rank of the $ i $-th individual. For two solutions with different nondomination ranks, if $ \text{irank} < \text{jrank} $, then the $ i $-th solution with the lower (better) rank is preferred.

### Algorithm .2 Fast-Nondominated-Sort(P)
# Fast Nondominated Sort Algorithm

The Fast Nondominated Sort algorithm is a crucial component of multi-objective optimization. It efficiently sorts a population of solutions into different nondominated fronts.

## Algorithm 2: Fast Nondominated Sort

Given a set of solutions $ P $:

1. For each $ p \in P $:
   1. $ S_p = \emptyset $
   2. $ n_p = 0 $
   3. For each $ q \in P $:
      1. If $ x_p \prec x_q $ then $\quad $ % if $ p $ dominates $ q $
         - $ S_p = S_p \cup \{q\} $
      2. Else if $ x_q \prec x_p $ then
         - $ n_p = n_p + 1 $
   4. If $ n_p = 0 $ then:
      1. $ rank_p = 1 $
      2. $ F_1 = F_1 \cup \{p\} 4

2. Let $ i = 1 $

3. While $ F_i \neq \emptyset $:
   1. $ Q = \emptyset $
   2. For each $ p \in F_i $:
      1. For each $ q \in S_p $:
         1. $ n_q = n_q - 1 $
         2. If $ n_q = 0 $ then:
            - $ rank_q = i + 1 $
            - $ Q = Q \cup \{q\} $

   3. $ i = i + 1 $
   4. $ F_i = Q $

4. Output: $ F = (F_1, F_2, \ldots) $ = fast-nondominated-sort($ P $), nondominated fronts of $ P $.



In order to identify solutions of the first nondominated front in a population of size \( NP \), each solution can be compared with every other solution in the population to determine if it is dominated. At this stage, all individuals in the first nondominated level in the population are found. To find the individuals in the second nondominated level, the solutions of the first level are discounted temporarily, and the above procedure is repeated to find the third and higher levels or ranks of nondomination.


```python
def dominates(p, q):
    """
    Check if solution p dominates solution q.
    :param p: List or tuple representing the solution p.
    :param q: List or tuple representing the solution q.
    :return: True if p dominates q, otherwise False.
    """
    return all(pi <= qi for pi, qi in zip(p, q)) and any(pi < qi for pi, qi in zip(p, q))

def fast_nondominated_sort(P):
    """
    Perform fast nondominated sorting on a population P.
    :param P: List of solutions, where each solution is a list or tuple.
    :return: List of fronts, where each front is a list of solutions.
    """
    F = []  # List of fronts
    S = []  # List of solutions that dominate others
    n = []  # Count of dominated solutions for each solution
    rank = [0] * len(P)  # Rank for each solution

    # Initialize the structures
    for _ in P:
        S.append([])
        n.append(0)

    # Fast nondominated sorting
    for i, p in enumerate(P):
        for j, q in enumerate(P):
            if dominates(p, q):
                S[i].append(q)  # p dominates q
            elif dominates(q, p):
                n[i] += 1  # q dominates p

        if n[i] == 0:
            rank[i] = 1  # p is in the first front
            if len(F) < 1:
                F.append([])
            F[0].append(p)

    i = 0
    while i < len(F):
        Q = []  # New front
        for p in F[i]:
            for q in S[P.index(p)]:
                n[P.index(q)] -= 1  # Reduce the count of dominated solutions
                if n[P.index(q)] == 0:
                    rank[P.index(q)] = i + 2  # Set rank of q
                    if len(F) <= i + 1:
                        F.append([])  # Create a new front if necessary
                    F[i + 1].append(q)  # Add q to the new front
        i += 1

    return F

# Example usage
P = [
    (1, 2),
    (2, 1),
    (1, 3),
    (3, 1),
    (2, 2),
    (3, 3)
]

fronts = fast_nondominated_sort(P)

# Output the fronts
for i, front in enumerate(fronts):
    print(f"Front {i + 1}: {front}")
```

## Crowding Distance Assignment Approach

In multi-objective optimization, when two solutions have the same nondominated rank (say $ \text{irank} = \text{jrank} $), we need another criterion for selecting the better solution. A natural choice is to estimate the density of solutions surrounding a particular solution in the population.

## Definition 22: Crowding Distance

Let $ f_1 \$ and $ f_2 $ be two objective functions in a multi-objective optimization problem, and let $ x $, $ x_i $, and $ x_j $ be members of a nondominated list of solutions. Furthermore, $ x_i $ and $ x_j $ are the nearest neighbors of $ x $ in the objective spaces.

The crowding distance (CD) of a trial solution $ x $ in a nondominated set depicts the perimeter of a hypercube formed by its nearest neighbors (i.e., $ x_i $ and $ x_j $) at the vertices in the fitness landscapes.

### Crowding Distance Computation

To compute the crowding distance, we need to sort the population according to each objective function value in ascending order. Let the population consist of $ |I| $ individuals, which are ranked according to their fitness values from small to large. Let $ f_{ki} = f_k(x_i) $ denote the $ k $th objective function of the $ i $th individual, where $ i = 1, \ldots, |I| $ and $ k = 1, \ldots, K $.

For individuals in the intermediate population set $ \{2, \ldots, |I| - 1\} $, the crowding distance $ CD_i $ is calculated as follows:

$$
CD_i = \sum_{k=1}^{K} \frac{f_{k,i+1} - f_{k,i-1}}{f_{k,\text{max}} - f_{k,\text{min}}}, \quad i = 2, \ldots, |I| - 1
$$

where

- $f_{k,\text{max}} = \max \{f_{k,1}, \ldots, f_{k,|I|}\} $
- $ f_{k,\text{min}} = \min \{f_{k,1}, \ldots, f_{k,|I|}\} $

The distance values of boundary individuals are assigned as:

$$
CD_1 = CD_{|I|} = \infty
$$

### Crowded-Comparison Operator

The crowded-comparison operator guides the selection process at various stages of the algorithm toward a uniformly spread-out Pareto-optimal front. Every population has two attributes: nondomination rank and crowding distance. By combining the nondominated rank and the crowded distance, the crowded-comparison operator is defined by partial order $ \prec_n $ as follows:

$$
i \prec_n j \quad \text{if} \quad (\text{irank} < \text{jrank}) \quad \text{or} \quad (\text{irank} = \text{jrank} \, \text{and} \, CD_i > CD_j)
$$

That is, between two solutions with differing nondomination ranks, the solution with the lower (better) rank is preferred. Otherwise, if both solutions belong to the same front, the solution that is located in a less crowded region is preferred.



## Algorithm .3: Crowding-Distance Assignment $ I $ [25]

Given a set of individuals $ I $, this algorithm calculates the crowding distance $ CD $ for each individual based on each objective function $ f_j $. The crowding distance helps in selecting solutions that are more uniformly spread-out across the Pareto front.

### Algorithm Steps

1. Set $ l = |I| $, the total number of individuals in $ I $.
2. For each individual $ i $, set $ CD_i = 0 $ to initialize crowding distances.
3. For each objective $ j $:
   - Sort the individuals $ I $ in ascending order based on objective $ j $.
   - Set the boundary crowding distances: $ CD_1 = CD_l = \infty $.
   - For each individual $ i = 2 $ to $ l - 1 $:
     - Calculate the crowding distance for the $ i $-th individual as:
       $$
       CD_i = CD_i + \frac{f_{j, i+1} - f_{j, i-1}}{f_{j, \text{max}} - f_{j, \text{min}}}
       $$
       where:
       - $ f_{j, i+1} $ and $ f_{j, i-1} $ are the objective values of the nearest neighbors of $ i $ for objective $ j $.
       - $ f_{j, \text{max}} $ and $ f_{j, \text{min}} $ are the maximum and minimum values of objective $ j $ across all individuals in $ I $.

4. End the process once crowding distances for all individuals and objectives are computed.

### LaTeX Notation for Implementation

This algorithm assigns higher crowding distance values to boundary solutions to ensure they remain in the Pareto front, thereby promoting a more evenly distributed selection of solutions across the objectives.

$$
CD_1 = CD_{|I|} = \infty
$$


```python
def crowding_distance_assignment(population, objectives):
    """
    Calculates the crowding distance for each individual in a population.

    Parameters:
    - population: List of individuals, where each individual is represented by a list of objective values.
    - objectives: Number of objectives for each individual.

    Returns:
    - crowding_distances: List of crowding distances for each individual in the population.
    """
    num_individuals = len(population)
    crowding_distances = [0] * num_individuals  # Initialize crowding distances

    # Calculate crowding distance for each objective
    for obj_index in range(objectives):
        # Sort population based on the current objective
        population_sorted = sorted(range(num_individuals), key=lambda x: population[x][obj_index])

        # Set crowding distance of boundary solutions to infinity
        crowding_distances[population_sorted[0]] = float('inf')
        crowding_distances[population_sorted[-1]] = float('inf')

        # Calculate crowding distance for intermediate solutions
        min_obj = population[population_sorted[0]][obj_index]
        max_obj = population[population_sorted[-1]][obj_index]
        obj_range = max_obj - min_obj if max_obj > min_obj else 1  # Avoid division by zero

        for i in range(1, num_individuals - 1):
            next_individual = population[population_sorted[i + 1]][obj_index]
            prev_individual = population[population_sorted[i - 1]][obj_index]
            crowding_distances[population_sorted[i]] += (next_individual - prev_individual) / obj_range

    return crowding_distances

# Example usage
population = [
    [1, 2],
    [2, 3],
    [3, 1],
    [4, 5],
    [5, 0]
]
objectives = 2

crowding_distances = crowding_distance_assignment(population, objectives)
for i, distance in enumerate(crowding_distances):
    print(f"Individual {i + 1}: Crowding Distance = {distance}")
```

## Noisy Multiobjective Optimization

In real-world applications, optimization problems often include multiple conflicting objectives and uncertainties. These uncertainties make the optimization problem "noisy" or "imprecise," and handling such **noisy multiobjective optimization problems** requires special techniques.

## Definitions and Concepts

Optimization problems containing both uncertainty and multiple objectives are known as **uncertain multiobjective optimization problems**. When the uncertainty in objectives is represented as stochastic noise, these problems are also termed as **noisy multiobjective optimization problems**.

### Interval Representation of Noisy Objectives

The uncertain or noisy objectives $ f_i(x, c_i) $ can be represented in an interval form:

$$
f_i(x, c_i) = \left[ \underline{f}_i(x, c_i), \overline{f}_i(x, c_i) \right]
$$

where $ c_i $ is a noise vector associated with the objective $ f_i $.

### Types of Uncertainty

Uncertainty can be categorized into two types:

1. **Aleatory Uncertainty** (Noise): An inherent property of the system or environment that cannot be reduced. It is represented by stochastic noise.

2. **Epistemic Uncertainty** (Imprecision): This arises due to incomplete knowledge or information about the system.

### Representation of Uncertain Multiobjective Optimization Problem

The noisy multiobjective optimization problem can be formulated as follows:

$$
\max/\min \ f = \left[ f_1(x, c_1), \dots, f_m(x, c_m) \right]^T,
$$

subject to:

$$
c_i = [c_{i1}, \dots, c_{il}]^T, \quad c_{ij} = [\underline{c}_{ij}, \overline{c}_{ij}]
$$

where:

- $ x $ is the $ d $-dimensional decision variable,
- $ X $ is the decision space of $ x $,
- $ f_i(x, c_i) = [\underline{f}_i(x, c_i), \overline{f}_i(x, c_i)] $ is the $ i $-th objective function interval,
- $ m $ is the number of objectives (with $ m \geq 3 $).

### Interval Order Relation

Consider two objective intervals $ f(x_1) = [\underline{f}(x_1), \overline{f}(x_1)] $ and $ f(x_2) = [\underline{f}(x_2), \overline{f}(x_2)] $. Their order relations are defined as:

1. **Less than or equal (Interval Order)**:
   $$
   f(x_1) \leq_I f(x_2) \Leftrightarrow \underline{f}(x_1) \leq \underline{f}(x_2) \ \text{and} \ \overline{f}(x_1) \leq \overline{f}(x_2)
   $$

2. **Greater than or equal (Interval Order)**:
   $$
   f(x_1) \geq_I f(x_2) \Leftrightarrow \underline{f}(x_1) \geq \underline{f}(x_2) \ \text{and} \ \overline{f}(x_1) \geq \overline{f}(x_2)
   $$

3. **Strictly less (Interval Order)**:
   $$
   f(x_1) <_I f(x_2) \Leftrightarrow \underline{f}(x_1) \leq \underline{f}(x_2), \ \overline{f}(x_1) \leq \overline{f}(x_2), \ \text{and} \ f(x_1) \neq f(x_2)
   $$

4. **Strictly greater (Interval Order)**:
   $$
   f(x_1) >_I f(x_2) \Leftrightarrow \underline{f}(x_1) \geq \underline{f}(x_2), \ \overline{f}(x_1) \geq \overline{f}(x_2), \ \text{and} \ f(x_1) \neq f(x_2)
   $$

5. **Non-comparable**:
   $$
   f(x_1) \sim f(x_2) \Leftrightarrow \text{neither} \ f(x_1) \leq_I f(x_2) \ \text{nor} \ f(x_2) \leq_I f(x_1)
   $$

## Pareto Concepts for Noisy Multiobjective Optimization

In noisy multiobjective optimization, solutions are not fixed points but approximation sets due to the uncertain nature of objectives. The Pareto dominance relationship between solutions involves comparison of sets rather than individual points.

### Outperformance (Dominance) Relation

In this context, the dominance between two approximation sets, \( A \) and \( B \), involves comparing the internally nondominated objectives within each set.

The goal of an evolutionary multiobjective optimization (EMO) algorithm is to approximate the true Pareto front as closely as possible. To compare EMO algorithms, the **nondominated sets** produced by them are evaluated to determine their approximation quality to the true Pareto front.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import random

# Define the number of solutions and objectives
num_solutions = 100
num_objectives = 2  # For 2-objective visualization

# Generate random solutions with noisy objectives
solutions = []
for _ in range(num_solutions):
    # Random decision variable x in range [0, 1]
    x = random.uniform(0, 1)
    
    # Objective functions with added noise
    f1 = x + random.uniform(-0.1, 0.1)  # Adding noise to f1
    f2 = (1 - x)**2 + random.uniform(-0.1, 0.1)  # Adding noise to f2
    
    # Store as tuple with intervals (lower and upper bounds due to noise)
    solutions.append({
        "x": x,
        "f1": (f1 - 0.05, f1 + 0.05),  # Interval for f1
        "f2": (f2 - 0.05, f2 + 0.05)   # Interval for f2
    })

# Identify Pareto front based on interval dominance
pareto_solutions = []
for sol in solutions:
    dominated = False
    for other in solutions:
        if (sol["f1"][0] >= other["f1"][1] and sol["f2"][0] >= other["f2"][1]) and (sol["f1"] != other["f1"] or sol["f2"] != other["f2"]):
            dominated = True
            break
    if not dominated:
        pareto_solutions.append(sol)

# Prepare data for plotting
f1_values = [sol["f1"][0] for sol in solutions]
f2_values = [sol["f2"][0] for sol in solutions]
pareto_f1 = [sol["f1"][0] for sol in pareto_solutions]
pareto_f2 = [sol["f2"][0] for sol in pareto_solutions]

# Plot all solutions and highlight Pareto front
plt.figure(figsize=(10, 6))
sns.scatterplot(x=f1_values, y=f2_values, label="All Solutions", color="blue")
sns.scatterplot(x=pareto_f1, y=pareto_f2, label="Pareto Front", color="red")
plt.xlabel("Objective f1")
plt.ylabel("Objective f2")
plt.title("Noisy Multiobjective Optimization with Pareto Front")
plt.legend()
plt.show()
```

## Noisy Multiobjective Optimization: Set Dominance Relations

In noisy multiobjective optimization, we deal with uncertain objectives, requiring various set dominance relations to compare approximation sets. Below are the formal definitions for dominance, weak dominance, strict dominance, and other related concepts.

## Dominance for Sets

**Definition.24 (Dominance for Sets):**  
An approximation set $ A $ is said to dominate another approximation set $ B $, denoted by $ A \succ B $, if every $ x_2 \in B $ is dominated by at least one $ x_1 \in A $ in all objectives. Formally:

- For minimization:
  $$
  A \succ B \iff \exists x_1 \in A, f(x_1) <_{I_N} f(x_2), \ \forall x_2 \in B
  $$

- For maximization:
  $$
  A \succ B \iff \exists x_1 \in A, f(x_1) >_{I_N} f(x_2), \ \forall x_2 \in B
  $$

Alternatively, we can write:
$$
A \succ B \iff N_D(A \cup B) = A \ \text{and} \ B \setminus N_D(A \cup B) \neq \emptyset
$$
where $ N_D(S) $ denotes the set of nondominated points in $ S $.

## Weak Dominance for Sets

**Definition.25 (Weak Dominance for Sets):**  
An approximation set $ A $ weakly dominates another set $ B $, denoted by $ A \succeq B $, if every $ x_2 \in B $ is weakly dominated by at least one $ x_1 \in A $ in all objectives. Formally:

- For minimization:
  $$
  A \succeq B \iff \exists x_1 \in A, f(x_1) \leq_{I_N} f(x_2), \ \forall x_2 \in B
  $$

- For maximization:
  $$
  A \succeq B \iff \exists x_1 \in A, f(x_1) \geq_{I_N} f(x_2), \ \forall x_2 \in B
  $$

Alternatively, we can write:
$$
A \succeq B \iff N_D(A \cup B) = A \ \text{and} \ A \neq B
$$
In this case, $ A $ weakly dominates $ B $ if all points in $ B $ are “covered” by those in $ A $, with at least one point in $ A $ not contained in $ B $.

## Strict Dominance for Sets

**Definition.26 (Strict Dominance for Sets):**  
An approximation set $ A $ strictly dominates another approximation set $ B $, denoted by $ A \gg B $, if every $ x_2 \in B $ is strictly dominated by at least one $ x_1 \in A $ in all objectives. Formally:

- For minimization:
  $$
  A \gg B \iff \exists x_1 \in A, f_i(x_1) <_{I_N} f_i(x_2), \ \forall x_2 \in B \ \text{and for all} \ i
  $$

- For maximization:
  $$
  A \gg B \iff \exists x_1 \in A, f_i(x_1) >_{I_N} f_i(x_2), \ \forall x_2 \in B \ \text{and for all} \ i
  $$

Alternatively, we can write:
$$
A \gg B \iff N_D(A \cup B) = A \ \text{and} \ B \cap N_D(A \cup B) = \emptyset
$$

## Better Relation

**Definition.27 (Better):**  
An approximation set $ A $ is better than another approximation set $ B $, denoted as $ A \succcurlyeq B $, and is defined as:

- For minimization:
  $$
  A \succcurlyeq B \iff \exists x_1 \in A, f(x_1) \leq_{I_N} f(x_2), \ \forall x_2 \in B \ \text{and} \ A \neq B
  $$

- For maximization:
  $$
  A \succcurlyeq B \iff \exists x_1 \in A, f(x_1) \geq_{I_N} f(x_2), \ \forall x_2 \in B \ \text{and} \ A \neq B
  $$

From this relation, we can conclude:
$$
A \succeq B \implies A \succcurlyeq B \ \vee \ A = B
$$

### Summary of Dominance Relations

The relationship between dominance relations can be summarized as follows:
$$
A \gg B \implies A \succ B \implies A \succcurlyeq B \implies A \succeq B
$$


```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Define the noisy objective functions
def objective_1(x):
    """First objective function with noise."""
    return x ** 2 + random.uniform(-1, 1)  # Noisy quadratic function

def objective_2(x):
    """Second objective function with noise."""
    return (x - 2) ** 2 + random.uniform(-1, 1)  # Noisy shifted quadratic function

def generate_population(size):
    """Generate a random population of solutions."""
    return [random.uniform(-5, 5) for _ in range(size)]

def evaluate_population(population):
    """Evaluate the objectives for a given population."""
    objectives = [(objective_1(x), objective_2(x)) for x in population]
    return objectives

def non_dominated_sort(objectives):
    """Perform non-dominated sorting on the objectives."""
    dominated = []
    nondominated = []
    
    for i, obj1 in enumerate(objectives):
        is_dominated = False
        for j, obj2 in enumerate(objectives):
            if i != j:
                # Check if obj1 dominates obj2
                if (obj1[0] >= obj2[0] and obj1[1] > obj2[1]) or (obj1[0] > obj2[0] and obj1[1] >= obj2[1]):
                    is_dominated = True
                    break
        if is_dominated:
            dominated.append(obj1)
        else:
            nondominated.append(obj1)
    
    return nondominated, dominated

def evolutionary_algorithm(pop_size, generations):
    """Run the evolutionary algorithm."""
    population = generate_population(pop_size)
    pareto_fronts = []

    for gen in range(generations):
        objectives = evaluate_population(population)
        nondominated, _ = non_dominated_sort(objectives)
        pareto_fronts.append(nondominated)

        # Generate new population (simple random selection for demo)
        population = generate_population(pop_size)

    return pareto_fronts

def plot_results(pareto_fronts):
    """Plot the results of the optimization."""
    for front in pareto_fronts:
        x_values = [obj[0] for obj in front]
        y_values = [obj[1] for obj in front]
        sns.scatterplot(x=x_values, y=y_values)

    plt.title('Noisy Multiobjective Optimization Results')
    plt.xlabel('Objective 1')
    plt.ylabel('Objective 2')
    plt.xlim(-5, 10)
    plt.ylim(-5, 10)
    plt.grid()
    plt.show()

# Parameters
population_size = 100
number_of_generations = 20

# Run the evolutionary algorithm and plot results
pareto_fronts = evolutionary_algorithm(population_size, number_of_generations)
plot_results(pareto_fronts)
```
