$$
\textbf{Machine Learning - Man Vs Machine}
$$

Machine learning (ML) is a subset of artificial intelligence, which builds a mathematical model based on sample data, known as “training data,” in order to make predictions or decisions without being explicitly programmed to perform the task.

A good definition given by Mitchell [181] is:

$$
\text{A computer program is said to learn from experience } E \text{ with respect to some class of tasks } T \text{ and performance measure } P \text{ if its performance at tasks in } T, \text{ as measured by } P, \text{ improves with experience } E.
$$

In machine learning, neural networks, support vector machines, and evolutionary computation, we are usually given a training set and a test set. The training set consists of the union of the labeled set and the unlabeled set of examples available to machine learners. In comparison, the test set consists of examples never seen before.

Let 

$$
(X_l, Y_l) = \{(x_1, y_1), \ldots, (x_l, y_l)\}
$$

denote the labeled set, where \(x_i \in \mathbb{R}^D\) is the \(i\)-th \(D\)-dimensional data vector and \(y_i \in \mathbb{R}\) or \(y_i \in \{1, \ldots, M\}\) is the corresponding label of the data vector \(x_i\). In regression problems, \(y_i\) is the regression or fitting of \(x_i\). In classification problems, \(y_i\) is the corresponding class label of \(x_i\) among the \(M\) classes of targets. The labeled data \(x_i\), \(i = 1, \ldots, l\) are observed by the user, while \(y_i\), \(i = 1, \ldots, l\) are labeled by data labeling experts or supervisors. The unlabeled set is comprised of the data vectors and is denoted by 

$$
X_u = \{x_1, \ldots, x_u\}.
$$

Machine learning aims to establish a regressor or classifier through learning the training set, and then to evaluate the performance of the regressor or classifier through the test set.

According to the nature of training data, we can classify machine learning as follows:

$$
\textbf{1. Regular or Euclidean Structured Data Learning}
$$

- \textbf{Supervised Learning}: Given a training set consisting of labeled data 

$$
(X_{\text{train}}, Y_{\text{train}}) = \{(x_1, y_1), \ldots, (x_l, y_l)\},
$$

supervised learning learns a general rule that maps inputs to outputs. This is like a “teacher” or a supervisor (data labeling expert) giving a student a problem (finding the mapping relationship between inputs and outputs) and its solutions (labeled output data) and telling that student to figure out how to solve other, similar problems: finding the mapping from the features of unseen samples to their correct labels or target values in the future.

- \textbf{Unsupervised Learning}: In unsupervised learning, the training set consists of the unlabeled set only 

$$
X_{\text{train}} = X_u = \{x_1, \ldots, x_u\}.
$$

The main task of the machine learner is to find the solutions on its own (i.e., patterns, structures, or knowledge in unlabeled data). This is like giving a student a set of patterns and asking him or her to figure out the underlying motifs that generated the patterns.

- \textbf{Semi-supervised Learning}: Given a training set 

$$
(X_{\text{train}}, Y_{\text{train}}) = \{(x_1, y_1), \ldots, (x_l, y_l)\} \cup \{x_{l+1}, \ldots, x_{l+u}\}
$$

with \(l \ll u\), i.e., we are given a small amount of labeled data together with a large amount of unlabeled data. Semi-supervised learning falls between unsupervised learning (without any labeled training data) and supervised learning (with completely labeled training data). Depending on how the data is labeled, semi-supervised learning can be divided into the following categories:

  - \textbf{Self-training} is a semi-supervised learning using its own predictions to teach itself.
  - \textbf{Co-training} is a weakly semi-supervised learning for multi-view data using the co-training setting, and uses their own predictions to teach themselves.
  - \textbf{Active Learning} is a semi-supervised learning where the learner has some active or participatory role in determining which data points it will ask to be labeled by an expert or teacher.

- \textbf{Reinforcement Learning}: Training data (in the form of rewards and punishments) is given only as feedback to an artificial intelligence agent in a dynamic environment. This feedback between the learning system and the interaction experience is useful to improve performance in the task being learned. The machine learning based on data feedback is called reinforcement learning. Q-learning is a popular model-free reinforcement learning and learns a reward or punishment function (action-value functions, simply called Q-function).

- \textbf{Transfer Learning}: In many real-world applications, the data distribution changes or data are outdated, and thus it is necessary to apply transfer learning for considering transfer of knowledge from the source domain to the target domain. Transfer learning includes but is not limited to inductive transfer learning, transductive transfer learning, unsupervised transfer learning, multitask learning, self-taught transfer learning, domain adaptation, and EigenTransfer.

Graph machine learning is an irregular or non-Euclidean structured data learning, and learns the structure of a graph, called also graph construction, from training samples in semi-supervised and unsupervised learning cases.

$$
\textbf{6.2.1 Single-Objective Composite Optimization}
$$

In supervised machine learning, we are given the collection of a data set 

$$
\{(x_1, y_1), \ldots, (x_N, y_N)\},
$$

where \(x_i \in \mathbb{R}^d\) for each \(i \in \{1, \ldots, N\}\) represents the feature vector of a target, and the scalar \(y_i\) is a label indicating whether a corresponding feature vector belongs (\(y_i = 1\)) or not (\(y_i = -1\)) to a particular class (i.e., topic of interest).

With a set of examples \(\{(x_i, y_i)\}_{i=1}^N\), define the prediction function with respect to the \(i\)-th sample as

$$
h(w; x_i) = w \cdot x_i = \langle w, x_i \rangle = w^T x_i,
$$

where the predictor \(w = [w_1, \ldots, w_d]^T \in \mathbb{R}^d\) is a \(d\)-dimensional parameter vector of the machine learning model that should have good “generalization ability.”

Construct a machine learning program in which the performance of a prediction function \(\hat{y}_i = h(w; x_i)\) is measured by counting how often the program prediction \(f(x_i; w)\) differs from the correct prediction \(y_i\), i.e., \(\hat{y}_i \neq y_i\). Therefore, we need to search for a predictor function that minimizes the frequency of observed mispredictions, called the empirical risk:

$$
R_n(h) = \frac{1}{N} \sum_{i=1}^N 1[ h(x_i) \neq y_i ],
$$

where

$$
1[A] = \begin{cases}
1, & \text{if } A \text{ is true;} \\
0, & \text{otherwise.}
\end{cases}
$$

The optimization in machine learning seeks to design a classifier/predictor \(w\) so that \(\hat{y} = \langle w, x \rangle\) can provide the correct classification/prediction for any unknown input vector \(x\). The objective \(f(w; x)\) in machine learning can be represented as the following common form:

$$
f(w; x) = l(w; x) + h(w; x),
$$

where \(l(w; x)\) is the loss function and \(h(w; x)\) denotes the regularization function.

In this manner, the expected risk for a given \(w\) is the expected value of the loss function \(l(w; x)\) taken with respect to the distribution of \(x\) as follows [31]:

$$
\text{(Expected Risk)} \quad R(w) = \mathbb{E}\{ l(w; x) \}.
$$

Since the true distribution of \(x\) is often unknown, the expected risk is commonly estimated by the empirical risk:

$$
\text{(Empirical Risk)} \quad R_n(w) = \frac{1}{N} \sum_{i=1}^N l_i(w),
$$

where \(l_i(w) = l(w; x_i)\).

If we were to try to minimize the empirical risk only, the method would be referred to as Empirical Risk Minimization (ERM). However, the empirical risk usually underestimates the true/expected risk by the estimation error, and thus we add a regularization term \(h(w; x)\) to the composite function.

The corresponding prediction error is given by

$$
e_i(w) = e(w; x_i, y_i) = w^T x_i - y_i.
$$

Then one can write the empirical error as the average of the prediction errors for all \(N\) samples \(\{(x_i, y_i)\}_{i=1}^N\):

$$
\text{(Empirical Error)} \quad R_{\text{emp}}(w) = \frac{1}{N} \sum_{i=1}^N e_i(w).
$$

The loss function incurred by the parameter vector \(w\) can be represented as

$$
l(w) = \frac{1}{2N} \sum_{i=1}^N | e_i(w) |^2 = \frac{1}{2N} \sum_{i=1}^N (w^T x_i - y_i)^2.
$$

The above machine learning tasks such as classification and regression can be represented as solving a “composite” (additive) optimization problem:

$$
\min_w f(w) = l(w) + h(w) = \frac{1}{2N} \sum_{i=1}^N (w^T x_i - y_i)^2 + h(w),
$$

where the objective function \(f(x)\) is given by the sum of \(N\) component loss functions \(f_i(w) = | e_i(w) |^2 = (w^T x_i - y_i)^2\) and a possibly nonsmooth regularization function \(h(w)\). It is assumed that each component loss function \(f_i(w; x_i) : \mathbb{R}^d \times \mathbb{R}^d \to \mathbb{R}\) is continuously differentiable and the sum of \(N\) component loss functions \(f(x)\) is strongly convex, while the regularization function \(h : \mathbb{R}^d \to \mathbb{R}\) is proper, closed, and convex but not necessarily differentiable, and is used for preventing overfitting.

The composite optimization tasks exist widely in various artificial intelligence applications, e.g., image recognition, speech recognition, task allocation, text classification/processing, and so on.

To simplify notation, denote \(g_j(w_k) = g(w_k; x_j, y_j)\) and \(f_j(w_k) = f(w_k; x_j, y_j)\) hereafter. Optimization methods for machine learning fall into two broad categories [31]:

$$
\textbf{• Stochastic Method:}
$$

The prototypical stochastic method is the stochastic gradient (SG) method [214] defined by

$$
w_{k+1} \leftarrow w_k - \alpha_k \nabla f_{ik}(w_k),
$$

for all \(k \in \{1, 2, \ldots\}\), the index \(i_k\), corresponding to the sample pair \((x_{i_k}, y_{i_k})\), is chosen randomly from \(\{1, \ldots, N\}\) and \(\alpha_k\) is a positive stepsize. Each iteration of this method involves only the computation of the gradient \(\nabla f_{ik}(w_k)\) corresponding to one sample.

$$
\textbf{• Batch Method:}
$$

The simplest form of batch method is the steepest descent algorithm:

$$
w_{k+1} \leftarrow w_k - \alpha_k \nabla R_n(w_k) = w_k - \alpha_k \frac{1}{N} \sum_{i=1}^N \nabla f_i(w_k).
$$

The steepest descent algorithm is also referred to as the gradient, batch gradient, or full gradient method.

The gradient algorithms are usually represented as

$$
w_{k+1} = w_k - \alpha_k g(w_k).
$$

Traditional gradient-based methods are effective for solving small-scale learning problems, but in the context of large-scale machine learning one of the core strategies of interest is the SG method proposed by Robbins and Monro [214]:

$$
g(w_k) \leftarrow \frac{1}{|S_k|} \sum_{i \in S_k} \nabla f_i(w_k),
$$

$$
w_{k+1} \leftarrow w_k - \alpha_k g(w_k),
$$

for all \(k \in \{1, 2, \ldots\}\) and \(S_k \subset \{1, \ldots, N\}\) is a small subset of sample pairs \((x_i, y_i)\) chosen randomly from \(\{1, \ldots, N\}\) and \(\alpha_k\) is a positive stepsize.

$$
\textbf{Algorithm 6.1 Stochastic Gradient (SG) Method [31]}
$$

1. \textbf{Input:} Sample pairs \((x_j, y_j)\), \(j = 1, \ldots, N\).
2. \textbf{Initialization:} Choose an initial iterate \(w_1\).
3. \textbf{For} \(k = 1, 2, \ldots\) \textbf{do}
    1. Generate a realization of the random variables \((x_i, y_i)\), where \(i \in S_k\) and \(S_k \subset \{1, \ldots, N\}\) is a small subset of samples.
    2. Compute a stepsize \(\alpha_k > 0\).
    3. Choose a stochastic gradient \(g(w_k) \leftarrow \frac{1}{|S_k|} \sum_{i \in S_k} \nabla f(w_k; x_i, y_i)\).
    4. Update the new iterate as \(w_{k+1} \leftarrow w_k - \alpha_k g(w_k)\).
    5. \textbf{Exit:} if \(w_{k+1}\) is converged.
4. \textbf{End for}
5. \textbf{Output:} \(\bar{w} = \frac{1}{k+1} \sum_{i=1}^{k+1} w_i\).

The SG iteration is very inexpensive, requiring only the evaluation of \(g(w_k, \xi_k)\). However, since SG generates noisy iterate sequences that tend to oscillate around minimizers during the optimization process, a natural idea is to compute a corresponding sequence of iterate averages:

$$
\bar{w}_{k+1} = \frac{1}{k+1} \sum_{i=1}^{k+1} w_i.
$$

where the averaged sequence \(\{\bar{w}_k\}\) has no effect on the computation of the SG iterate sequence \(\{w_k\}\). Iterate averages would automatically possess better convergence properties.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function and regularization
def loss_function(X, y, w):
    predictions = X.dot(w)
    errors = predictions - y
    return (1 / (2 * len(y))) * np.sum(errors ** 2)

def regularization(w, lambda_reg):
    return (lambda_reg / 2) * np.sum(w ** 2)

def composite_function(X, y, w, lambda_reg):
    return loss_function(X, y, w) + regularization(w, lambda_reg)

# Stochastic Gradient Descent (SGD)
def stochastic_gradient_descent(X, y, w, learning_rate, lambda_reg, num_epochs):
    N = len(y)
    for epoch in range(num_epochs):
        # Shuffle indices
        indices = np.random.permutation(N)
        for i in indices:
            xi = X[i, :].reshape(1, -1)
            yi = y[i]
            gradient = xi.T.dot(xi.dot(w) - yi) + lambda_reg * w
            w -= learning_rate * gradient
    return w

# Batch Gradient Descent (BGD)
def batch_gradient_descent(X, y, w, learning_rate, lambda_reg, num_epochs):
    N = len(y)
    for epoch in range(num_epochs):
        predictions = X.dot(w)
        errors = predictions - y
        gradient = X.T.dot(errors) / N + lambda_reg * w
        w -= learning_rate * gradient
    return w

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 2)
y = 4 + 3 * X[:, 0] + np.random.randn(100)

# Add bias term
X_b = np.c_[np.ones((100, 1)), X]

# Initialize weights
w_init = np.random.randn(X_b.shape[1])

# Parameters
learning_rate = 0.01
lambda_reg = 0.1
num_epochs = 1000

# Run Stochastic Gradient Descent
w_sgd = stochastic_gradient_descent(X_b, y, w_init.copy(), learning_rate, lambda_reg, num_epochs)
print("Weights from SGD:", w_sgd)

# Run Batch Gradient Descent
w_bgd = batch_gradient_descent(X_b, y, w_init.copy(), learning_rate, lambda_reg, num_epochs)
print("Weights from BGD:", w_bgd)

# Plotting results
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], y, color='blue', label='Data')
plt.plot(X[:, 0], X_b.dot(w_sgd), color='red', label='SGD Fit')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('SGD Fit')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], y, color='blue', label='Data')
plt.plot(X[:, 0], X_b.dot(w_bgd), color='green', label='BGD Fit')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('BGD Fit')
plt.legend()

plt.tight_layout()
plt.show()
```

# Gradient Aggregation Methods in Machine Learning

## 6. Machine Learning

Consider a common single-objective convex optimization problem:

$$
\min_{x \in \mathbb{R}^d} \left\{ F(x) = f(x) + h(x) \right\}
$$

where the first term \( f(x) \) is smooth and the second term \( h(x) \) is possibly nonsmooth, which allows for the modeling of constraints.

In many applications in optimization, pattern recognition, signal processing, and machine learning, \( f(x) \) has an additional structure:

$$
f(x) = \frac{1}{N} \sum_{i=1}^{N} f_i(x)
$$

which is the average of a number of convex functions \( f_i(x) \).

When \( N \) is large and when a solution with low to medium accuracy is sufficient, it is common to perform classical stochastic methods, especially stochastic gradient descent (SGD) for solving the optimization problem. The SGD method was first proposed in the 1951 seminal work by Robbins and Monro. SGD selects an index \( i \in \{1, \dots, N\} \) uniformly at random, and then updates \( x \) using a stochastic estimate \( \nabla f_i(x) \) of \( \nabla f(x) \).

SG algorithms only use the current gradient information, while **gradient aggregation** reuses or revises previously computed gradients to achieve improved convergence rates and lower variance. Some common gradient aggregation methods are:

### 6.2.1 Stochastic Average Gradient (SAG)

SAG replaces iterate averages:

$$
g(w_k) = \frac{1}{N} \sum_{i=1}^{N} \left( \nabla f_j(w_k) - \nabla f_j(w_{k-1}) + \nabla f_i(w[i]) \right)
$$

Here, \( \nabla f_j(w_k) \) is chosen at random from the current gradient, \( \nabla f_j(w_{k-1}) \) is chosen from the past gradient, and \( \nabla f_i(w[i]) \) represents the latest iterate at which \( \nabla f_i \) was evaluated.

### 6.2.2 Stochastic Variance Reduced Gradient (SVRG)

SVRG improves SGD with a variance reduction (VR) technique:

$$
g̃_j \leftarrow \nabla f_{ij}(w̃_j) - \nabla f_{ij}(w_k) - \frac{1}{N} \sum_{i=1}^{N} \nabla f_i(w_k)
$$

### 6.2.3 Stochastic Average Gradient Aggregation (SAGA)

SAGA combines ideas from SAG and SVRG. The stochastic gradient vector is:

$$
g_k \leftarrow \nabla f_j(w_k) - \nabla f_j(w[j]) + \frac{1}{N} \sum_{i=1}^{N} \nabla f_i(w[i])
$$

where \( j \in \{1, \dots, N\} \) is chosen at random, and \( \nabla f_j(w[j]) \) and \( \nabla f_i(w[i]) \) are set by:

$$
\nabla f_j = \frac{\partial f(w_k; x_j, y_j)}{\partial w_k}
$$

$$
\nabla f_j(w[j]) = \frac{\partial f(w; x_j, y_j)}{\partial w} \bigg|_{w=w[j]}
$$

$$
\nabla f_i(w[i]) = \frac{\partial f(w; x_i, y_i)}{\partial w} \bigg|_{w=w[i]}
$$

### Algorithm 6.3: SAGA for Minimizing Empirical Risk

SAGA method for minimizing an empirical risk \( E_n \) is shown in Algorithm 6.3. Theoretical convergence rates for SAGA in the strongly convex case are better than for SAG and SVRG. SAGA is also applicable to non-strongly convex problems without modification.


# Algorithm 6.2: SVRG Method for Minimizing an Empirical Risk $ R_n $ 

1. **Input**: Sample pairs $ (x_j , y_j) $, $ j = 1, \dots, N $, update frequency $ m $, and learning rate $ \alpha $.
2. **Initialization**: Choose an initial iterate $ w_1 \in \mathbb{R}^d $.
3. **For** $ k = 1, 2, \dots $, do:
   1. Compute the gradient:
   $$ \nabla_i(w_k) = \frac{\partial f(w_k; y_i)}{\partial w} ; x_i $$
   
4. Compute the batch gradient:
   $$ R_n(w_k) = \frac{1}{N} \sum_{i=1}^{N} \nabla f_i(w_k) $$
   
5. Initialize:
   $$ \tilde{w}_1 \leftarrow w_k $$

6. **For** $ j = 1, \dots, m $, do:
   1. Choose $ i_j $ uniformly from $ \{1, \dots, N\} $.
   2. Set:
   $$ g̃_j \leftarrow \nabla f_{i_j}(\tilde{w}_j) - (\nabla f_{i_j}(w_k) - R_n(w_k)) $$
   3. Update:
   $$ \tilde{w}_{j+1} \leftarrow \tilde{w}_j - \alpha g̃_j $$

7. **End for**

8. **Options**:
   - (a) Set:
   $$ w_{k+1} = \tilde{w}_{m+1} $$
   
   - (b) Set:
   $$ w_{k+1} = \frac{1}{m} \sum_{j=1}^{m} \tilde{w}_{j+1} $$
   
   - (c) Choose \( j \) uniformly from \( \{1, \dots, m\} \) and set:
   $$ w_{k+1} = \tilde{w}_{j+1} $$

9. **Exit**: If $ w_{k+1} $ is converged.
10. **End for**
11. **Output**: $ w = w_{k+1} $


```python
import numpy as np

def svrg(X, y, w_init, learning_rate=0.01, update_frequency=10, num_epochs=100):
    """
    Implements the SVRG (Stochastic Variance Reduced Gradient) method for minimizing an empirical risk.
    
    Parameters:
        X (numpy.ndarray): Feature matrix of shape (N, d), where N is the number of samples and d is the number of features.
        y (numpy.ndarray): Labels vector of shape (N,).
        w_init (numpy.ndarray): Initial weights of shape (d,).
        learning_rate (float): The step size or learning rate (alpha) for gradient updates.
        update_frequency (int): The number of iterations for which the variance-reduced gradient is computed.
        num_epochs (int): The number of epochs for which the algorithm runs.
    
    Returns:
        w (numpy.ndarray): The optimized weights of shape (d,).
    """
    
    N, d = X.shape  # Number of samples (N) and features (d)
    w = w_init.copy()  # Initialize weights
    w_avg = np.zeros_like(w)  # Averaged weights for variance reduction
    
    for epoch in range(num_epochs):
        # Compute full batch gradient at w (denoted R_n(w_k))
        full_gradient = np.mean([np.dot(X[i], w) - y[i] for i in range(N)], axis=0)
        
        # Initialize w_tilde as w (w_tilde_1 <- w_k)
        w_tilde = w.copy()
        
        for j in range(update_frequency):
            # Select random index i_j
            i_j = np.random.randint(0, N)
            
            # Compute gradients for the chosen sample
            grad_i_j_tilde = np.dot(X[i_j], w_tilde) - y[i_j]
            grad_i_j_w = np.dot(X[i_j], w) - y[i_j]
            
            # Variance reduced gradient (SVRG update)
            gradient = grad_i_j_tilde - grad_i_j_w + full_gradient
            
            # Update w_tilde
            w_tilde -= learning_rate * gradient
            
        # Option (a): Set w_{k+1} = w_tilde_{m+1}
        w = w_tilde.copy()
    
    return w

# Example usage
np.random.seed(42)
X = np.random.rand(100, 10)  # 100 samples, 10 features
y = np.random.rand(100)      # Random labels
w_init = np.zeros(10)        # Initial weights

# Run SVRG
w_svrg = svrg(X, y, w_init, learning_rate=0.01, update_frequency=10, num_epochs=100)

print("Optimized weights:", w_svrg)
```

7. & \quad \text{Choose} \ j \ \text{uniformly in} \ \{1, \dots, N\}. \\
8. & \quad \text{Compute} \ \nabla f_j(w_k) = \frac{\partial f(w_k; y_j)}{\partial w}; x_j \\
9. & \quad \text{Set} \ g_k \leftarrow \nabla f_j(w_k) - \nabla f_j(w[j]) + \frac{1}{N} \sum_{i=1}^{N} \nabla f_i(w[i]) \\
10. & \quad \text{Store} \ \nabla f_j(w[j]) \leftarrow \nabla f_j(w_k). \\
11. & \quad \text{Set} \ w_{k+1} \leftarrow w_k - \alpha g_k \\
12. & \textbf{Exit:} \ \text{If} \ w_{k+1} \ \text{is converged}. \\
13. & \textbf{End for} \\
14. & \textbf{Output:} \ w = w_{k+1}.
\end{aligned}


```python
import numpy as np

# SAGA algorithm implementation
def saga(X, y, w_init, learning_rate, num_epochs):
    """
    Implements the SAGA algorithm for minimizing an empirical risk function.
    
    Parameters:
    X: numpy.ndarray
        Feature matrix of size (N, d) where N is the number of samples and d is the number of features.
    y: numpy.ndarray
        Target values of size (N,).
    w_init: numpy.ndarray
        Initial weight vector of size (d,).
    learning_rate: float
        Learning rate (step size) for gradient updates.
    num_epochs: int
        Number of epochs to run the algorithm.
        
    Returns:
    w: numpy.ndarray
        The final weight vector after running the SAGA algorithm.
    """
    
    # Initialize parameters
    N, d = X.shape
    w = w_init.copy()
    w_old = np.zeros_like(w)
    grad_memory = np.zeros((N, d))  # Store gradient history for each sample
    avg_grad = np.zeros(d)  # Store the average gradient
    total_grad = np.zeros(d)  # Total gradient sum

    # Compute initial gradient for each sample and store it
    for i in range(N):
        grad_memory[i] = X[i] * (X[i].dot(w_init) - y[i])
        avg_grad += grad_memory[i]
    
    avg_grad /= N

    # Training loop
    for epoch in range(num_epochs):
        for i in range(N):
            # Randomly select an index
            j = np.random.randint(N)

            # Compute the current gradient for the selected sample
            grad_j = X[j] * (X[j].dot(w) - y[j])

            # SAGA update rule
            w -= learning_rate * (grad_j - grad_memory[j] + avg_grad)

            # Update the gradient memory and average gradient
            avg_grad += (grad_j - grad_memory[j]) / N
            grad_memory[j] = grad_j

    return w

# Example usage
N = 100  # Number of samples
d = 10   # Number of features

# Generating random data
np.random.seed(42)
X = np.random.randn(N, d)
y = np.random.randn(N)
w_init = np.random.randn(d)

# Set hyperparameters
learning_rate = 0.01
num_epochs = 100

# Run SAGA
w_saga = saga(X, y, w_init, learning_rate, num_epochs)

print("Final weights after SAGA:", w_saga)
```

### Coordinate Descent Methods

Coordinate descent methods (CDMs) solve optimization problems by minimizing the objective function along coordinate directions iteratively. Each iteration updates one variable while keeping the others fixed.

The basic update rule for CDMs is:

$$
w_{k+1} \leftarrow w_k - \alpha_k \nabla_{i_k} f(w_k) e_{i_k}
\tag{6.2.25}
$$

where:

$$
\nabla_{i_k} f(w_k) = \frac{\partial f(w_k)}{\partial w_{i_k}}
\tag{6.2.26}
$$

- \( w_k \) is the current parameter vector at iteration \( k \).
- \( i_k \) is the index of the coordinate being updated.
- \( e_{i_k} \) is the unit coordinate vector with 1 at the \( i_k \)-th position and 0 elsewhere.
- \( \alpha_k \) is the step size (learning rate).
- \( \nabla_{i_k} f(w_k) \) is the partial derivative of the objective function with respect to \( w_{i_k} \).

### Example 1: Simple Quadratic Function

Consider a simple quadratic objective function:

$$
f(w) = \frac{1}{2} \lVert w \rVert^2 = \frac{1}{2} \sum_{i=1}^{d} w_i^2
$$

The partial derivative is:

$$
\nabla_{i_k} f(w_k) = w_{i_k}
$$

The update rule becomes:

$$
w_{k+1,i} =
\begin{cases}
w_{k,i} - \alpha_k w_{i_k}, & i = i_k \\
w_{k,i}, & \text{otherwise}
\end{cases}
$$

### Categories of Coordinate Descent Methods

There are two main categories of CDMs based on how the coordinate \( i_k \) is selected:

1. **Cyclic Coordinate Descent**: The coordinates are updated in a fixed cyclic order.
2. **Stochastic Coordinate Descent**: The coordinates are chosen randomly at each iteration.

In the stochastic approach, the coordinates are selected either:
- Randomly with replacement in each iteration, or
- Through a random reordering of the indices after each set of \( d \) steps.

### Randomized Coordinate Descent

For minimizing a convex and differentiable function:

$$
\min_x f(x) \quad \text{where} \quad x = [x_1, \dots, x_N]^T \in \mathbb{R}^N
\tag{6.2.27}
$$

The coordinate descent update rule is:

$$
x_{k+1} = x_k - \alpha_k U_{i_k}^T \nabla_{i_k} f(x_k)
$$

where:
- \( U_{i_k} \) is the unit matrix corresponding to the \( i_k \)-th coordinate.
- \( \nabla_{i_k} f(x) \) is the partial derivative with respect to \( x_{i_k} \).

### Block Coordinate Descent

In block coordinate descent, the decision vector \( x \in \mathbb{R}^N \) is divided into non-overlapping blocks:

$$
x = \left[ x^{(1)}, \dots, x^{(n)} \right]^T
\tag{6.2.28}
$$

where \( x^{(i)} \in \mathbb{R}^{N_i} \) and \( N = \sum_{i=1}^n N_i \).

The block coordinate descent update rule is:

$$
x_{k+1} = x_k - \alpha_k U_i^T \nabla_i f(x_k)
\tag{6.2.31}
$$

where \( U_i \) corresponds to the block \( i \), and \( \nabla_i f(x) \) is the partial derivative with respect to the block \( x^{(i)} \).

### Example 6.2: Block Gradient Computation

Let \( S_1 = \{1, 2\} \), and \( U_1 = [e_1, e_2] \). Then:

$$
x^{(1)} = \left[ x_1, x_2 \right]^T, \quad \nabla_1 f(x) = \left[ f'(x_1), f'(x_2) \right]^T
$$

where \( f'(x_j) = \frac{\partial f(x)}{\partial x_j} \) for \( j = 1, 2 \).

### Accelerated Coordinate Descent

Nesterov proposed an accelerated coordinate descent method (ACDM) for convex minimization in 2012. It is theoretically efficient for solving large-scale problems but is computationally expensive.

The optimization problem is given as:

$$
\min_x F(x) = f(x) + h(x)
\tag{6.2.33}
$$

where:
- \( f(x) \) is a smooth convex objective function separable in blocks \( x^{(i)} \).
- \( h(x) \) is a convex regularizer separable in blocks \( x^{(i)} \).

### APPROX Method

Fercoq and Richtárik (2015) proposed the APPROX randomized block coordinate descent method, which is accelerated, parallel, and proximal, to solve the optimization problem.



$\textbf{Algorithm 6.4: APPROX Coordinate Descent Method }$

$$
\text{Input:}
$$

$$
\text{Initialization:} \, \tilde{z}_0 \in \mathbb{R}^N, \, \theta_0 = \frac{n}{\tau}, \, u_0 = 0
$$

$$
\text{For} \, k = 0, 1, \dots
$$

1. Generate a random set of blocks \( S_k \sim \hat{S} \).

2. Set $ u_{k+1} \leftarrow u_k $, $ \tilde{z}_{k+1} \leftarrow \tilde{z}_k $.

3. For each block \( i \in S_k \):

   $$
   t_k^{(i)} = \arg\max_{t \in \mathbb{R}^{N_i}} \langle \nabla_i f(\theta_k u_k + \tilde{z}_k), t \rangle + \frac{\tau}{2} \|t\|^2 + h_i(\tilde{z}_k^{(i)})
   $$

4. Update $ \tilde{z}_{k+1}^{(i)} $:

   $$
   \tilde{z}_{k+1}^{(i)} \leftarrow \tilde{z}_k^{(i)} + t_k^{(i)}
   $$

5. Update $ u_{k+1}^{(i)} $:

   $$
   u_{k+1}^{(i)} \leftarrow u_k^{(i)} - \frac{1 - \theta_k}{\tau \theta_k} t_k^{(i)}
   $$

6. Update $ \theta_{k+1} $:

   $$
   \theta_{k+1} = \frac{\theta_k}{2} \left( \sqrt{4 + 4 \theta_k^2} - \theta_k \right)
   $$

$$
\text{Output:} \, x = \theta_k^2 u_{k+1} + \tilde{z}_{k+1}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def approx_coordinate_descent(f, grad_f, h, z0, n_blocks, tau, max_iter=1000, epsilon=1e-5):
    """
    Implements the APPROX coordinate descent method with performance tracking and plotting.
    
    Parameters:
    f: function
        Objective function f(x).
    grad_f: function
        Gradient of the objective function ∇f(x).
    h: function
        Regularization function h(x).
    z0: np.array
        Initial vector z0 ∈ R^N.
    n_blocks: int
        Number of blocks for the block decomposition.
    tau: float
        Step size parameter.
    max_iter: int
        Maximum number of iterations.
    epsilon: float
        Convergence tolerance.
        
    Returns:
    x: np.array
        The optimized solution.
    performance_metrics: dict
        Dictionary with performance metrics.
    """
    
    N = len(z0)  # Dimension of the problem
    z = z0.copy()  # Initialize z
    u = np.zeros_like(z0)  # Initialize u as zero vector
    theta = n_blocks / tau  # Initialize theta
    
    # Performance tracking
    objective_values = []
    norm_gradients = []

    for k in range(max_iter):
        S_k = np.random.choice(N, size=1, replace=False)  # Random set of blocks
        u_k = u.copy()  # Keep a copy of u
        z_k = z.copy()  # Keep a copy of z
        
        for i in S_k:
            # Compute the coordinate gradient step tk
            grad_f_i = grad_f(theta * u_k + z_k, i)
            t_k = np.argmax(np.dot(grad_f_i, grad_f_i.T) + (tau / 2) * np.linalg.norm(grad_f_i) ** 2 + h(z_k[i]))
            
            # Update z and u
            z_k[i] += t_k
            u_k[i] -= (1 - theta) / (tau * theta) * t_k
        
        # Update theta
        theta_next = (theta / 2) * (np.sqrt(4 + 4 * theta ** 2) - theta)
        
        # Check convergence
        if np.linalg.norm(z_k - z) < epsilon:
            break
        
        z = z_k  # Update z
        u = u_k  # Update u
        theta = theta_next  # Update theta
        
        # Track performance metrics
        objective_values.append(f(theta * u + z))
        norm_gradients.append(np.linalg.norm(grad_f(theta * u + z)))
    
    # Final output
    x = theta ** 2 * u + z
    
    # Plot performance metrics
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(objective_values, label='Objective Value', color='blue')
    plt.xlabel('Iteration')
    plt.ylabel('Objective Value')
    plt.title('Objective Value Over Iterations')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(norm_gradients, label='Norm of Gradient', color='orange')
    plt.xlabel('Iteration')
    plt.ylabel('Norm of Gradient')
    plt.title('Norm of Gradient Over Iterations')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    return x, {
        'objective_values': objective_values,
        'norm_gradients': norm_gradients
    }

# Example functions and parameters
def f(x):
    # Example objective function: Quadratic function
    return 0.5 * np.dot(x, x)

def grad_f(x, i):
    # Example gradient function: Gradient of quadratic function
    grad = np.zeros_like(x)
    grad[i] = x[i]
    return grad

def h(x):
    # Example regularization function: L1 regularization
    return np.sum(np.abs(x))

# Initialize parameters
N = 50  # Dimension
z0 = np.random.randn(N)
n_blocks = N
tau = 0.1

# Run the APPROX coordinate descent method
x, metrics = approx_coordinate_descent(f, grad_f, h, z0, n_blocks, tau)

# Display the results
print("Optimized Solution:", x)
print("Final Objective Value:", f(x))
print("Performance Metrics:", metrics)
```

### Benchmark Functions for Single-Objective Optimization

The test of reliability, efficiency, and validation of an optimization algorithm is frequently carried out by using a chosen set of common benchmark or test functions. There have been many benchmark or test functions reported in the literature. Ideally, benchmark functions should have diverse properties when testing new algorithms in an unbiased way. On the other hand, there is a serious phenomenon called the “curse of dimensionality” in many optimization methods. This implies that the optimization performance deteriorates quickly as the dimensionality of the search space increases. The reasons for this phenomenon appear to be twofold:

1. The solution space of a problem often increases exponentially with the problem dimension, and more efficient search strategies are required to explore all promising regions within a given time budget.
2. The characteristics of a problem may change with the scale. Because an increase in scale may result in worsening of the features of an optimization problem, a previously successful search strategy may no longer be capable of finding the optimal solution.

To solve large-scale problems, it is necessary to provide a suite of benchmark functions for large-scale numerical optimization.

#### Optimization in Machine Learning

**Definition 6.1 (Separable Function)** Given \( x = [x_1, \dots, x_n]^T \). A function \( f(x) \) is a separable function if and only if
$$
\arg \min_{(x_1, \dots, x_n)} f(x_1, \dots, x_n) = (\arg \min_{x_1} f(x_1, \dots), \dots, \arg \min_{x_n} f(\dots, x_n)),
$$
where \( f(\dots, x_i, \dots) \) denotes that its decision variable is just \( x_i \) and other variables are constants with respect to \( x_i \).

That is to say, a function of \( n \) decision variables is separable if it can be rewritten as a sum of \( n \) functions of just one variable, i.e.,
$$
f(x_1, \dots, x_n) = f(x_1) + \cdots + f(x_n).
$$
If a function \( f(x) \) is separable, its parameters \( x_i \) are said to be independent. A separable function can be easily optimized, as it can be decomposed into a number of subproblems, each of which involves only one decision variable while treating all others as constants.

**Definition 6.2 (Nonseparable Function)** A nonseparable function \( f(x) \) is called an \( m \)-nonseparable function if at most \( m \) (where \( m < n \)) of its decision variables \( x_i \) are not independent. A nonseparable function \( f(x) \) is known as a fully nonseparable function if any two of its parameters \( x_i \) are not independent.

A nonseparable function has at least partly coupled or dependent decision variables.

The definitions of separability provide us a measure of the difficulty of different optimization problems based on which a spectrum of benchmark problems can be designed. In general, separable problems are considered to be easiest, while the fully nonseparable ones are usually the most difficult.

# Majorization-Minimization Algorithms

In the era of big data, the rapid development in data acquisition techniques and the growth of computing power have led to large-scale problems with a tremendous amount of data and variables. This poses challenges to traditional algorithms \cite{87}. 

The Majorization-Minimization (MM) algorithm, initially discussed by Hunter and Lange \cite{125}, and first enunciated in the 1970s \cite{197}, is a powerful tool for tackling difficult optimization problems. The essence of MM algorithms is to simplify the optimization problem by modifying the objective function, making it easier to explore the solution space.

### Key Concepts

The main idea behind MM algorithms is to replace a difficult optimization problem with a simpler one. The simplicity is achieved by:

- Avoiding large matrix inversions,
- Linearizing an optimization problem,
- Separating the parameters of an optimization problem,
- Dealing with equality and inequality constraints gracefully, or
- Turning a non-differentiable problem into a smooth problem.

MM algorithms have a close relationship with the well-known Expectation-Maximization (EM) algorithm, which is widely used in computational statistics \cite{278}.

### MM Algorithm Framework

Consider the minimization problem:

$$
\hat{\theta} = \arg\min_{\theta \in \Theta} g(\theta)
$$

where \( g(\theta) \) is a difficult-to-manipulate objective function, and \( \Theta \) is a subset of some Euclidean space. For example, \( g(\theta) \) might be non-differentiable or non-convex.

Let \( \theta^{(m)} \) be a fixed value of the parameter \( \theta \) at the \( m \)-th iteration. Define \( g(\theta | \theta^{(m)}) \) as a real-valued function of \( \theta \) whose form depends on \( \theta^{(m)} \). Instead of working directly with \( g(\theta) \), we consider an easier surrogate function \( h(\theta | \theta^{(m)}) \) at \( \theta^{(m)} \).

#### Definition 6.3 (Majorizer)

The surrogate function \( h(\theta | \theta^{(m)}) \) is said to be a majorizer of \( g(\theta) \) if:

$$
h(\theta | \theta^{(m)}) \geq g(\theta) \quad \text{for all } \theta
$$

and

$$
h(\theta^{(m)} | \theta^{(m)}) = g(\theta^{(m)})
$$

The majorization step involves finding the majorizer \( h(\theta | \theta^{(m)}) \) of \( g(\theta) \). The function \( h(\theta | \theta^{(m)}) \) lies above \( g(\theta) \) and is tangent to \( g(\theta) \) at \( \theta^{(m)} \). If \( \theta \) minimizes \( h(\theta | \theta^{(m)}) \), then:

$$
g(\theta^{(m)}) = h(\theta^{(m)} | \theta^{(m)}) \geq h(\theta^{(m+1)} | \theta^{(m)}) \geq g(\theta^{(m+1)})
$$

This implies that the sequence \( \{ g(\theta^{(m)}) \} \) is non-increasing, pushing \( g(\theta) \) toward its minimum.

The function \( h(\theta | \theta^{(m)}) \) is said to minorize \( g(\theta) \) at \( \theta^{(m)} \) if \( -h(\theta | \theta^{(m)}) \) majorizes \( -g(\theta) \).

#### Minimization Step

The minimization step updates \( \theta \) as:

$$
\theta^{(m+1)} = \arg\min_{\theta} h(\theta | \theta^{(m)})
$$

This step ensures the numerical stability of the MM algorithm by relaxing the difficult loss function \( g(\theta) \) to a simpler surrogate function \( h(\theta | \theta^{(m)}) \) that is easier to minimize.

### Example: Generalized Eigenvalue Problem

Consider the generalized eigenvalue problem for a matrix pencil \( (A, B) \):

$$
\lambda = \frac{x^T A x}{x^T B x}
$$

The variational formulation for the maximum generalized eigenvalue \( \lambda_{\text{max}}(A, B) \) is:

$$
\lambda_{\text{max}}(A, B) = \max_{x} \frac{x^T A x}{x^T B x}
$$

subject to \( x^T B x = 1 \). 

To address the sparse version of this problem:

$$
\max_{x} \frac{x^T A x}{x^T B x}
$$

subject to \( x^T B x \leq 1 \) and \( \|x\|_0 \leq k \)

we can use a regularized version:

$$
\max_{x} x^T A x - \rho \|x\|_0
$$

subject to \( x^T B x \leq 1 \)

Here, \( \rho > 0 \) is a regularization parameter. By approximating the non-convex \( \|x\|_0 \) norm with a smooth function, the problem can be converted into a simpler convex optimization problem.

### Applications

MM algorithms are extensively used in various fields including:

- Fully visible Boltzmann machine estimation
- Linear mixed model estimation
- Markov random field estimation
- Matrix completion and imputation
- Quantile regression estimation
- Support Vector Machine (SVM) estimation

For a comprehensive list of applications, see \cite{191}.

MM algorithms offer a robust framework for developing effective algorithms in machine learning, statistical estimation, and signal processing.



```python
import numpy as np
from scipy.optimize import minimize

# Generate synthetic data for demonstration
np.random.seed(0)
n, p = 100, 50  # dimensions
A = np.random.randn(n, p)
b = np.random.randn(n)
lambda_reg = 0.1

# Objective function: least-squares with L1 regularization
def objective(theta, A, b, lambda_reg):
    return 0.5 * np.linalg.norm(A @ theta - b)**2 + lambda_reg * np.linalg.norm(theta, 1)

# Majorization step: define the surrogate function
def majorizer(theta, A, b, lambda_reg):
    residual = A @ theta - b
    surrogate = lambda theta_new: 0.5 * np.linalg.norm(A @ theta_new - b)**2 + lambda_reg * np.linalg.norm(theta_new, 1)
    return surrogate

# Minimization step: optimize the surrogate function
def minimize_surrogate(surrogate, theta):
    result = minimize(surrogate, theta, method='L-BFGS-B')
    return result.x

# Initial guess for theta
theta_initial = np.zeros(p)

# MM algorithm
def majorization_minimization(A, b, lambda_reg, max_iter=10):
    theta = theta_initial
    for _ in range(max_iter):
        # Majorization step
        surrogate = majorizer(theta, A, b, lambda_reg)
        
        # Minimization step
        theta = minimize_surrogate(surrogate, theta)
    
    return theta

# Run MM algorithm
theta_opt = majorization_minimization(A, b, lambda_reg)

print("Optimized theta:", theta_opt)
```

## Majorization-Minimization Algorithms

In optimization, Majorization-Minimization (MM) algorithms are a class of iterative procedures used to tackle difficult optimization problems. They are designed to simplify the optimization process by replacing complex objective functions with simpler surrogate functions that are easier to minimize. 

### MM Algorithm Framework

Consider the minimization problem:

$$
\hat{\theta} = \arg \min_{\theta \in \Theta} g(\theta)
$$

where \( g(\theta) \) is a difficult-to-manipulate objective function. For the MM algorithm, we use an easier surrogate function \( h(\theta \mid \theta^{(m)}) \) that majorizes \( g(\theta) \) at the point \( \theta^{(m)} \). 

**Definition 6.3:** A surrogate function \( h(\theta \mid \theta^{(m)}) \) is a majorizer of \( g(\theta) \) if:

$$
h(\theta \mid \theta^{(m)}) \geq g(\theta) \text{ for all } \theta
$$

$$
h(\theta^{(m)} \mid \theta^{(m)}) = g(\theta^{(m)})
$$

**Definition 6.4:** The MM algorithm updates \( \theta \) by solving:

$$
\theta^{(m+1)} = \arg \min_{\theta \in \Theta} h(\theta \mid \theta^{(m)})
$$

The sequence \( \{\theta^{(m)}\} \) will have the property that the objective function values \( g(\theta^{(m)}) \) are monotonically decreasing:

$$
g(\theta^{(m+1)}) \leq h(\theta \mid \theta^{(m+1)}) \leq h(\theta^{(m)} \mid \theta^{(m)}) \leq g(\theta^{(m)})
$$

### Example: Weighted \( \ell_0 \) Minimization

Consider the weighted \( \ell_0 \) minimization problem:

$$
\min_{x \in \mathbb{R}^n} \|W x\|_0 \text{ subject to } y = A x
$$

where \( W \) is a weight matrix and \( \| \cdot \|_0 \) denotes the \( \ell_0 \) norm, which counts the number of non-zero entries.

To simplify this, we use a surrogate function related to the weighted \( \ell_1 \) minimization:

$$
\min_{x \in \mathbb{R}^n} \sum_{i=1}^p w_i |x_i| \text{ subject to } y = A x
$$

where \( w_i \) are weights.

**Surrogate Function:** If \( x_i^{(m)} \) are the current estimates, the surrogate function can be approximated by:

$$
\sum_{i=1}^n \left( |x_i^{(m)}| + \epsilon \log \left(1 + \frac{|x_i|}{\epsilon}\right) \right)
$$

where \( \epsilon \) is a small positive value.

**Algorithm Steps:**
1. **Initialization:**
   - Set the iteration count \( m \) to zero and initialize \( w_i^{(0)} = 1 \) for all \( i \).

2. **Solve Weighted \( \ell_1 \) Minimization:**
   - Solve:

   $$
   x^{(m)} = \arg \min_{x \in \mathbb{R}^n} \sum_{i=1}^n w_i^{(m)} |x_i| \text{ subject to } y = A x
   $$

3. **Update Weights:**
   - For each \( i \), update:

   $$
   w_i^{(m+1)} = \frac{1}{|x_i^{(m)}| + \epsilon}
   $$

4. **Check Convergence:**
   - Terminate on convergence or when \( m \) reaches a maximum number of iterations \( m_{\text{max}} \). Otherwise, increment \( m \) and repeat.

### Example: Total Variation Minimization

For a 2D array \( (x_{i,j}) \), consider the total variation (TV) norm minimization:

$$
\min_{x \in \mathbb{R}^n} \|D x\|_{\text{TV}} \text{ subject to } y = A x
$$

where \( D x \) denotes the forward differences.

**Algorithm Steps:**
1. **Initialization:**
   - Set \( m = 0 \) and initialize weights \( w_{i,j}^{(0)} \).

2. **Solve Weighted TV Minimization:**
   - Solve:

   $$
   x^{(m)} = \arg \min_{x \in \mathbb{R}^n} \sum_{i,j} w_{i,j}^{(m)} \| (D x)_{i,j} \| \text{ subject to } y = A x
   $$

3. **Update Weights:**
   - For each \( (i,j) \), update:

   $$
   w_{i,j}^{(m+1)} = \frac{1}{\| (D x)_{i,j} \| + \epsilon}
   $$

4. **Check Convergence:**
   - Terminate on convergence or when \( m \) reaches a maximum number of iterations \( m_{\text{max}} \). Otherwise, increment \( m \) and repeat.

### Remarks

- **Stationary Points:** The MM algorithm guarantees convergence to a stationary point, but not necessarily to a global minimum.
- **Divergence:** If the limit point does not exist, the MM sequence may diverge.

The MM algorithm framework is widely applicable in fields such as machine learning, statistical estimation, and signal processing.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the total variation norm
def total_variation(x, shape):
    D = np.diff(x.reshape(shape), axis=0)  # Forward differences
    return np.sum(np.abs(D))

def solve_tv_minimization(A, y, shape):
    from scipy.linalg import lstsq
    x = lstsq(A, y, lapack_driver='gelsy')[0]
    return x

# Initial data
A = np.array([[1, 2], [3, 4]])  # Example matrix
y = np.array([5, 6])            # Example vector
shape = (2, 2)                  # Shape of the image
x = np.zeros(np.prod(shape))   # Initial guess
max_iter = 100
convergence_threshold = 1e-6
errors = []

# MM algorithm
for iteration in range(max_iter):
    x_old = x.copy()
    x = solve_tv_minimization(A, y, shape)
    tv_norm = total_variation(x, shape)
    error = tv_norm - total_variation(x_old, shape)
    errors.append(error)
    
    if error < convergence_threshold:
        break

# Plotting the error
plt.plot(errors)
plt.title('Convergence of Total Variation Minimization')
plt.xlabel('Iteration')
plt.ylabel('Objective Error')
plt.show()

print('Final x:', x)
```

## Boosting and Probably Approximately Correct Learning

Valiant and Kearns introduced the concepts of weak learning and strong learning. A learning algorithm with an error rate less than \( \frac{1}{2} \) (i.e., its accuracy is slightly better than random guessing) is called a weak learning algorithm. A learning algorithm with very high accuracy and polynomial time complexity is called a strong learning algorithm. The equivalence problem in the probably approximately correct (PAC) learning model questions whether a weak learning algorithm that performs only slightly better than random guessing can be upgraded to a strong learning algorithm.

In 1990, Schapire constructed a polynomial-time algorithm to prove that a weak learning algorithm that is slightly better than random guessing can be upgraded to a strong learning algorithm. This was the original boosting algorithm. Freund and Schapire later improved the boosting algorithm, proposing AdaBoost (Adaptive Boosting).

**Boosting** combines many weak classifiers to create a powerful ensemble. Boosting involves creating a series of base classifiers by sampling subsets of the data and then combining these classifiers to improve performance.

Let \( x_i \) be an instance and \( X = \{ x_1, \ldots, x_N \} \) be a domain (instance space). Let \( D_s \) be a fixed probability distribution over the instance space \( X \), and \( D_t \) be a distribution over the set of training examples \( X \times Y \). In most machine learning settings, it is assumed that \( D_s = D_t^X \), meaning the instance distribution and the target distribution are the same over the instance space.

A concept \( c \) over \( X \) is a subset of \( X \) with unknown labels and can be represented as a Boolean function \( c: X \to \{0, 1\} \), where \( c(x) = 1 \) indicates \( x \) is a positive example, and \( c(x) = 0 \) indicates \( x \) is a negative example.

**Definition 6.5 (Agree, Consistent):** A representation \( h \) and an example \( (x, y) \) agree if \( h(x) = y \); otherwise, they disagree. Let \( S = \{ (x_1, y_1), \ldots, (x_N, y_N) \} \) be a labeled set of instances. A concept \( c \) is consistent with \( S \) if \( c(x_i) = y_i \) for all \( 1 \leq i \leq N \); otherwise, it is inconsistent.

For \( M \) target classes where each class is represented by \( C_i \), the task of a classifier is to assign the input sample \( x \) to one of the \( M+1 \) classes, where the \( (M+1) \)th class indicates rejection. If \( h_1, \ldots, h_K \) are \( K \) classifiers, each providing a classification result \( h_i(x) \), the combined result \( H(x) \) is determined by majority voting.

**Majority Voting Method:**
1. Define a binary function to represent the number of votes:
   $$
   V_k(x \in C_i) = \begin{cases} 
   1 & \text{if } h_k(x) = i, \; i \in \{1, \ldots, M\} \\
   0 & \text{otherwise}
   \end{cases}
   $$

2. Sum the votes from all \( K \) classifiers for each \( C_i \):
   $$
   V_H(x \in C_i) = \sum_{k=1}^K V_k(x \in C_i), \; i = 1, \ldots, M
   $$

3. Determine the combined result \( H(x) \):
   $$
   H(x) = \begin{cases} 
   i & \text{if } i \in \{1, \ldots, M\} \text{ and } V_E(x \in C_i) \geq \alpha \cdot K \\
   M + 1 & \text{otherwise}
   \end{cases}
   $$
   where \( \alpha \) is a user-defined threshold that controls the confidence in the final decision.

**Definition 6.6 (Polynomially Evaluatable):** A representation class \( C \) over \( X \) is polynomially evaluatable if there is a polynomial-time (possibly randomized) evaluation algorithm \( A \) that, given a representation \( c \in C \) and a domain point \( x \in X \), runs in time polynomial in \( |c| \) and \( |x| \) and decides if \( x \in c(x) \).

A strong learning algorithm is polynomially evaluatable and produces a representation \( h \) that is consistent with a given sample \( S \). Designing a strong learning algorithm directly can be challenging, so boosting frameworks are used to enhance weak learners.

A weak learner is a classifier that performs slightly better than random guessing. Schapire showed that by training additional classifiers on filtered data, performance can be significantly improved. Specifically:
1. \( h_2 \) is learned on a new sample of \( N \) points, half of which are misclassified by \( h_1 \).
2. \( h_3 \) is learned on \( N \) points where \( h_1 \) and \( h_2 \) disagree.
3. The boosted classifier is \( h_B = \text{Majority Vote}(h_1, h_2, h_3) \), which performs better than a coinflip with high probability.

**Additive Logistic Regression Model:**
$$
\log \frac{P(Y = 1 \mid x)}{1 - P(Y = 1 \mid x)} = \beta_0 + \beta_1 x_1 + \cdots + \beta_p x_p
$$
where \( \frac{P(Y = 1 \mid x)}{1 - P(Y = 1 \mid x)} \) is the odds ratio, and \( \frac{P(Y = 1 \mid x)}{1 - P(Y = 1 \mid x)} = \frac{e^{F(x)}}{1 + e^{F(x)}} \).

**Lemma 6.2:** Minimizing the exponential criterion
$$
J(F) = \mathbb{E}\left\{ e^{-yF(x)} \right\}
$$
is achieved at
$$
F(x) = \frac{1}{2} \log \frac{P(y = 1 \mid x)}{P(y = -1 \mid x)}
$$
Thus,
$$
\frac{e^{F(x)}}{P(y = 1 \mid x)} = e^{-F(x)} + e^{F(x)}
$$
$$
\frac{e^{-F(x)}}{P(y = -1 \mid x)} = e^{F(x)} + e^{-F(x)}
$$

**LogitBoost Algorithm (two classes):**
1. **Input:** weights \( w_i = \frac{1}{N} \), \( i = 1, \ldots, N \), \( F(x) = 0 \), and probability estimates \( p(x_i) = \frac{1}{2} \).
2. For \( m = 1 \) to \( M \):
   - Compute the working response and weight:
     $$
     z_i = p(x_i) \left(y_i - p(x_i)\right)
     $$
     $$
     w_i = p(x_i) \left(1 - p(x_i)\right)
     $$
   - Fit the function \( f_m(x) \) by weighted least squares regression of \( z_i \) to \( x_i \) using weights \( w_i \).
   - Update:
     $$
     F(x) \leftarrow F(x) + \frac{1}{2} f_m(x)
     $$
     $$
     p(x) \leftarrow \frac{e^{F(x)}}{e^{F(x)} + e^{-F(x)}}
     $$
3. **Output:** The classifier \( \text{sign}[F(x)] = \text{sign} \left[ \sum_{m=1}^M f_m(x) \right] \).


```python
import numpy as np
from scipy.special import expit as sigmoid

class LogitBoost:
    def __init__(self, M=100):
        self.M = M  # Number of boosting iterations

    def fit(self, X, y):
        N, _ = X.shape
        self.weights = np.ones(N) / N  # Initialize weights
        self.F = np.zeros(N)  # Initialize F(x)
        self.p = np.ones(N) / 2  # Initialize probability estimates

        for _ in range(self.M):
            # Compute working response and weight
            z = self.p * (y - self.p)
            w = self.p * (1 - self.p)

            # Fit the function fm(x) by weighted least squares regression
            self.fm = self._fit_regression(X, z, w)

            # Update F(x)
            self.F += 0.5 * self.fm
            self.p = sigmoid(self.F)

    def _fit_regression(self, X, z, w):
        # Weighted least squares regression
        X_w = X * np.sqrt(w[:, np.newaxis])
        z_w = z * np.sqrt(w)
        
        # Solve for the regression coefficients
        coeffs = np.linalg.lstsq(X_w, z_w, rcond=None)[0]
        return X @ coeffs

    def predict(self, X):
        # Prediction based on the learned F(x)
        F_x = self.F
        return np.sign(F_x)

# Example usage
if __name__ == "__main__":
    from sklearn.datasets import make_classification
    from sklearn.metrics import accuracy_score

    # Create a synthetic dataset
    X, y = make_classification(n_samples=100, n_features=10, n_informative=5, n_classes=2, random_state=42)
    y = 2 * y - 1  # Convert to {-1, 1}

    # Train LogitBoost model
    model = LogitBoost(M=50)
    model.fit(X, y)

    # Predict and evaluate
    y_pred = model.predict(X)
    print("Accuracy:", accuracy_score(y, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit as sigmoid
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

class LogitBoost:
    def __init__(self, M=100):
        self.M = M  # Number of boosting iterations

    def fit(self, X, y):
        N, _ = X.shape
        self.weights = np.ones(N) / N  # Initialize weights
        self.F = np.zeros(N)  # Initialize F(x)
        self.p = np.ones(N) / 2  # Initialize probability estimates
        self.loss_history = []  # To track loss over iterations

        for _ in range(self.M):
            # Compute working response and weight
            z = self.p * (y - self.p)
            w = self.p * (1 - self.p)

            # Fit the function fm(x) by weighted least squares regression
            self.fm = self._fit_regression(X, z, w)

            # Update F(x)
            self.F += 0.5 * self.fm
            self.p = sigmoid(self.F)

            # Record the loss
            loss = np.mean(np.log(1 + np.exp(-y * self.F)))
            self.loss_history.append(loss)

    def _fit_regression(self, X, z, w):
        # Weighted least squares regression
        X_w = X * np.sqrt(w[:, np.newaxis])
        z_w = z * np.sqrt(w)
        
        # Solve for the regression coefficients
        coeffs = np.linalg.lstsq(X_w, z_w, rcond=None)[0]
        return X @ coeffs

    def predict(self, X):
        # Prediction based on the learned F(x)
        F_x = self.F
        return np.sign(F_x)

    def plot_loss(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(self.M), self.loss_history, marker='o')
        plt.title('LogitBoost Loss Convergence')
        plt.xlabel('Iteration')
        plt.ylabel('Loss')
        plt.grid(True)
        plt.show()

    def plot_decision_boundary(self, X, y):
        plt.figure(figsize=(10, 5))

        # Create a grid of points
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                             np.arange(y_min, y_max, 0.02))

        # Predict the class for each point in the grid
        Z = self.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        # Plot decision boundary
        plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
        plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.RdYlBu)
        plt.title('LogitBoost Decision Boundary')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.grid(True)
        plt.show()

# Example usage
if __name__ == "__main__":
    # Create a synthetic dataset
    # Ensure that informative + redundant + repeated <= total features
    X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, random_state=42)
    y = 2 * y - 1  # Convert to {-1, 1}

    # Train LogitBoost model
    model = LogitBoost(M=50)
    model.fit(X, y)

    # Plot loss convergence
    model.plot_loss()

    # Plot decision boundary
    model.plot_decision_boundary(X, y)

    # Predict and evaluate
    y_pred = model.predict(X)
    print("Accuracy:", accuracy_score(y, y_pred))
```
