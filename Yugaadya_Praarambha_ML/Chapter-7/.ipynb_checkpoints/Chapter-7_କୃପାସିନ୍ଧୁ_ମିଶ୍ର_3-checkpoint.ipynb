# Maximum Likelihood Estimation (MLE)

## 8.3 Parameter Estimation

In this section, we explore how to use probability distributions to model uncertainty in both the observation process and the parameters of our predictors. We'll see how the likelihood function relates to loss functions in empirical risk minimization, and how priors relate to regularization.

## 8.3.1 Maximum Likelihood Estimation

### Core Concept

Maximum Likelihood Estimation (MLE) defines a function of parameters that enables us to find a model that fits the data well. The estimation problem focuses on the **likelihood function**, or more precisely, its negative logarithm.

### Mathematical Foundation

For data represented by a random variable $x$ and a family of probability densities $p(x | \theta)$ parametrized by $\theta$, the **negative log-likelihood** is:

$$L_x(\theta) = -\log p(x | \theta)$$

**Key Points:**
- The notation $L_x(\theta)$ emphasizes that parameter $\theta$ is varying while data $x$ is fixed
- Often simplified to $L(\theta)$ when the random variable is clear from context
- $p(x | \theta)$ models the uncertainty of data for a given parameter setting

### Interpretation

The likelihood function allows us to:
1. Express preferences about different parameter settings $\theta$
2. Choose the setting that more "likely" generated the data
3. Find the most likely parameter $\theta$ for the observed data

## Supervised Learning Setting

Consider the supervised learning case with pairs $(x_1, y_1), \ldots, (x_N, y_N)$ where:
- $x_n \in \mathbb{R}^D$ (feature vectors)
- $y_n \in \mathbb{R}$ (labels)

**Goal:** Construct a predictor that takes feature vector $x_n$ as input and produces prediction $y_n$

We specify the conditional probability distribution of labels given examples for parameter setting $\theta$.

## Example 8.4: Gaussian Likelihood

### Assumptions

1. **Observation uncertainty:** Independent Gaussian noise with zero mean
   $$\varepsilon_n \sim \mathcal{N}(0, \sigma^2)$$

2. **Linear model:** Use $x_n^T \theta$ for prediction

3. **Gaussian likelihood:** For each example-label pair $(x_n, y_n)$:
   $$p(y_n | x_n, \theta) = \mathcal{N}(y_n | x_n^T \theta, \sigma^2)$$

### Independence and Identical Distribution

**Assumptions:**
- Examples $(x_1, y_1), \ldots, (x_N, y_N)$ are independent and identically distributed (i.i.d.)
- **Independent:** Likelihood factorizes into product of individual likelihoods
- **Identically distributed:** Each term shares the same distribution and parameters

### Dataset Likelihood

The likelihood for the entire dataset factorizes as:

$$p(Y | X, \theta) = \prod_{n=1}^N p(y_n | x_n, \theta)$$

where:
- $Y = \{y_1, \ldots, y_N\}$ (all labels)
- $X = \{x_1, \ldots, x_N\}$ (all features)

## Negative Log-Likelihood

For optimization purposes, we work with the negative log-likelihood:

$$L(\theta) = -\log p(Y | X, \theta) = -\sum_{n=1}^N \log p(y_n | x_n, \theta)$$

**Why negative log-likelihood?**
- Converts products to sums (easier optimization)
- Historical convention: maximize likelihood ≡ minimize negative log-likelihood
- Numerical optimization typically focuses on minimization

### Key Identity Used

$$\log(ab) = \log(a) + \log(b)$$

Therefore:
$$\log \prod_{n=1}^N p(y_n | x_n, \theta) = \sum_{n=1}^N \log p(y_n | x_n, \theta)$$

## Optimization Objective

To find the best parameter vector $\theta$ that explains the data $(x_1, y_1), \ldots, (x_N, y_N)$:

$$\hat{\theta} = \arg\min_{\theta} L(\theta) = \arg\min_{\theta} \left(-\sum_{n=1}^N \log p(y_n | x_n, \theta)\right)$$

## Important Note

Although $\theta$ appears on the right side of the conditioning in $p(y_n | x_n, \theta)$, it should **not** be interpreted as observed and fixed. The negative log-likelihood $L(\theta)$ is a function of $\theta$, which we optimize to find the best parameter values.

## Summary

Maximum Likelihood Estimation provides a principled framework for:
- Modeling uncertainty in data and parameters
- Finding optimal parameter settings
- Connecting probabilistic modeling with optimization
- Bridging the gap between loss functions and likelihood functions

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')

class GaussianMLE:
    """
    Maximum Likelihood Estimation for Linear Regression with Gaussian Likelihood
    
    Implements the concepts from Section 8.3.1:
    - Gaussian likelihood: p(y_n | x_n, θ) = N(y_n | x_n^T θ, σ²)
    - Negative log-likelihood: L(θ) = -∑ log p(y_n | x_n, θ)
    """
    
    def __init__(self, sigma=1.0):
        """
        Initialize the MLE estimator
        
        Args:
            sigma (float): Standard deviation of Gaussian noise
        """
        self.sigma = sigma
        self.theta = None
        self.fitted = False
        
    def gaussian_likelihood(self, y, x_theta):
        """
        Compute Gaussian likelihood p(y | x^T θ, σ²)
        
        Args:
            y (float): Observed label
            x_theta (float): Predicted value x^T θ
            
        Returns:
            float: Likelihood value
        """
        return norm.pdf(y, loc=x_theta, scale=self.sigma)
    
    def log_likelihood_single(self, y, x_theta):
        """
        Compute log-likelihood for a single example
        
        Args:
            y (float): Observed label
            x_theta (float): Predicted value x^T θ
            
        Returns:
            float: Log-likelihood value
        """
        return norm.logpdf(y, loc=x_theta, scale=self.sigma)
    
    def negative_log_likelihood(self, theta, X, y):
        """
        Compute negative log-likelihood L(θ) = -∑ log p(y_n | x_n, θ)
        
        Args:
            theta (np.ndarray): Parameter vector
            X (np.ndarray): Feature matrix (N x D)
            y (np.ndarray): Label vector (N,)
            
        Returns:
            float: Negative log-likelihood value
        """
        # Compute predictions: X @ theta
        predictions = X @ theta
        
        # Compute log-likelihood for each example
        log_likelihoods = [self.log_likelihood_single(y[i], predictions[i]) 
                          for i in range(len(y))]
        
        # Return negative sum (negative log-likelihood)
        return -np.sum(log_likelihoods)
    
    def fit(self, X, y, method='analytical'):
        """
        Fit the model using Maximum Likelihood Estimation
        
        Args:
            X (np.ndarray): Feature matrix (N x D)
            y (np.ndarray): Label vector (N,)
            method (str): 'analytical' or 'numerical' optimization
        """
        # Add bias term (intercept) if not present
        if X.shape[1] == 1 or not np.allclose(X[:, 0], 1):
            X = np.column_stack([np.ones(X.shape[0]), X])
        
        if method == 'analytical':
            # For Gaussian likelihood with fixed σ, MLE has closed-form solution
            # θ_ML = (X^T X)^(-1) X^T y (same as least squares!)
            self.theta = np.linalg.solve(X.T @ X, X.T @ y)
        
        elif method == 'numerical':
            # Use numerical optimization to minimize negative log-likelihood
            initial_theta = np.random.randn(X.shape[1])
            
            result = minimize(
                fun=self.negative_log_likelihood,
                x0=initial_theta,
                args=(X, y),
                method='BFGS'
            )
            
            self.theta = result.x
            
        self.fitted = True
        
    def predict(self, X):
        """
        Make predictions using fitted model
        
        Args:
            X (np.ndarray): Feature matrix
            
        Returns:
            np.ndarray: Predictions
        """
        if not self.fitted:
            raise ValueError("Model must be fitted before making predictions")
        
        # Add bias term if needed
        if X.shape[1] != len(self.theta):
            X = np.column_stack([np.ones(X.shape[0]), X])
            
        return X @ self.theta
    
    def likelihood_surface(self, X, y, theta_range):
        """
        Compute likelihood surface for visualization (2D parameter case)
        
        Args:
            X (np.ndarray): Feature matrix
            y (np.ndarray): Labels
            theta_range (tuple): Range for theta values
            
        Returns:
            tuple: (theta0_grid, theta1_grid, likelihood_grid)
        """
        theta0_vals = np.linspace(theta_range[0][0], theta_range[0][1], 50)
        theta1_vals = np.linspace(theta_range[1][0], theta_range[1][1], 50)
        
        theta0_grid, theta1_grid = np.meshgrid(theta0_vals, theta1_vals)
        likelihood_grid = np.zeros_like(theta0_grid)
        
        for i in range(len(theta0_vals)):
            for j in range(len(theta1_vals)):
                theta = np.array([theta0_grid[j, i], theta1_grid[j, i]])
                likelihood_grid[j, i] = np.exp(-self.negative_log_likelihood(theta, X, y))
        
        return theta0_grid, theta1_grid, likelihood_grid


def generate_synthetic_data(n_samples=100, true_theta=None, sigma=1.0, seed=42):
    """
    Generate synthetic data for testing MLE
    
    Args:
        n_samples (int): Number of samples
        true_theta (np.ndarray): True parameter values
        sigma (float): Noise standard deviation
        seed (int): Random seed
        
    Returns:
        tuple: (X, y, true_theta)
    """
    np.random.seed(seed)
    
    if true_theta is None:
        true_theta = np.array([2.0, -1.5])  # [intercept, slope]
    
    # Generate features
    x = np.random.uniform(-3, 3, n_samples)
    X = np.column_stack([np.ones(n_samples), x])  # Add bias term
    
    # Generate labels with Gaussian noise
    y_true = X @ true_theta
    noise = np.random.normal(0, sigma, n_samples)
    y = y_true + noise
    
    return X, y, true_theta


def demonstrate_mle():
    """
    Demonstrate Maximum Likelihood Estimation with examples
    """
    print("=" * 60)
    print("MAXIMUM LIKELIHOOD ESTIMATION DEMONSTRATION")
    print("=" * 60)
    
    # Generate synthetic data
    print("\n1. Generating Synthetic Data")
    print("-" * 30)
    
    X, y, true_theta = generate_synthetic_data(n_samples=100, sigma=1.0)
    print(f"True parameters: θ = {true_theta}")
    print(f"Data shape: X = {X.shape}, y = {y.shape}")
    
    # Fit MLE model
    print("\n2. Fitting MLE Model")
    print("-" * 30)
    
    mle = GaussianMLE(sigma=1.0)
    
    # Analytical solution
    mle.fit(X, y, method='analytical')
    print(f"MLE estimate (analytical): θ = {mle.theta}")
    
    # Numerical solution
    mle_numerical = GaussianMLE(sigma=1.0)
    mle_numerical.fit(X, y, method='numerical')
    print(f"MLE estimate (numerical):  θ = {mle_numerical.theta}")
    
    # Compare with true values
    error_analytical = np.linalg.norm(mle.theta - true_theta)
    error_numerical = np.linalg.norm(mle_numerical.theta - true_theta)
    print(f"Error (analytical): {error_analytical:.4f}")
    print(f"Error (numerical):  {error_numerical:.4f}")
    
    # Evaluate negative log-likelihood
    print("\n3. Negative Log-Likelihood Evaluation")
    print("-" * 30)
    
    nll_true = mle.negative_log_likelihood(true_theta, X, y)
    nll_estimated = mle.negative_log_likelihood(mle.theta, X, y)
    
    print(f"NLL at true parameters:      {nll_true:.4f}")
    print(f"NLL at estimated parameters: {nll_estimated:.4f}")
    print(f"Improvement: {nll_true - nll_estimated:.4f}")
    
    # Make predictions
    print("\n4. Making Predictions")
    print("-" * 30)
    
    X_test = np.column_stack([np.ones(5), np.linspace(-2, 2, 5)])
    y_pred = mle.predict(X_test)
    y_true_test = X_test @ true_theta
    
    print("Test predictions vs true values:")
    for i in range(len(y_pred)):
        print(f"  x = {X_test[i, 1]:5.2f}: pred = {y_pred[i]:6.3f}, true = {y_true_test[i]:6.3f}")
    
    return mle, X, y, true_theta


def visualize_mle_results(mle, X, y, true_theta):
    """
    Visualize MLE results
    """
    print("\n5. Visualization")
    print("-" * 30)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    # Plot 1: Data and fitted line
    x_vals = X[:, 1]  # Remove bias column for plotting
    x_plot = np.linspace(x_vals.min(), x_vals.max(), 100)
    X_plot = np.column_stack([np.ones(100), x_plot])
    
    y_pred_plot = mle.predict(X_plot)
    y_true_plot = X_plot @ true_theta
    
    axes[0].scatter(x_vals, y, alpha=0.6, label='Data')
    axes[0].plot(x_plot, y_pred_plot, 'r-', label='MLE fit', linewidth=2)
    axes[0].plot(x_plot, y_true_plot, 'g--', label='True function', linewidth=2)
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('y')
    axes[0].set_title('MLE Linear Regression')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Plot 2: Likelihood surface (for 2D parameter case)
    if len(true_theta) == 2:
        theta_range = [
            (true_theta[0] - 1, true_theta[0] + 1),
            (true_theta[1] - 1, true_theta[1] + 1)
        ]
        
        theta0_grid, theta1_grid, likelihood_grid = mle.likelihood_surface(X, y, theta_range)
        
        contour = axes[1].contour(theta0_grid, theta1_grid, likelihood_grid, levels=20)
        axes[1].plot(true_theta[0], true_theta[1], 'go', markersize=10, label='True θ')
        axes[1].plot(mle.theta[0], mle.theta[1], 'r*', markersize=15, label='MLE θ')
        axes[1].set_xlabel('θ₀ (intercept)')
        axes[1].set_ylabel('θ₁ (slope)')
        axes[1].set_title('Likelihood Surface')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
    
    # Plot 3: Residuals
    y_pred = mle.predict(X)
    residuals = y - y_pred
    
    axes[2].scatter(y_pred, residuals, alpha=0.6)
    axes[2].axhline(y=0, color='r', linestyle='--')
    axes[2].set_xlabel('Predicted values')
    axes[2].set_ylabel('Residuals')
    axes[2].set_title('Residual Plot')
    axes[2].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("Plots generated successfully!")


def compare_with_different_noise_levels():
    """
    Compare MLE performance with different noise levels
    """
    print("\n6. Effect of Noise Level on MLE")
    print("-" * 30)
    
    sigma_values = [0.5, 1.0, 2.0, 3.0]
    true_theta = np.array([2.0, -1.5])
    
    results = []
    
    for sigma in sigma_values:
        # Generate data with different noise levels
        X, y, _ = generate_synthetic_data(n_samples=200, true_theta=true_theta, sigma=sigma)
        
        # Fit MLE
        mle = GaussianMLE(sigma=sigma)
        mle.fit(X, y, method='analytical')
        
        # Compute error
        error = np.linalg.norm(mle.theta - true_theta)
        results.append((sigma, error, mle.theta))
        
        print(f"σ = {sigma:.1f}: θ_estimated = [{mle.theta[0]:6.3f}, {mle.theta[1]:6.3f}], "
              f"error = {error:.4f}")
    
    print(f"\nTrue parameters: θ = [{true_theta[0]:6.3f}, {true_theta[1]:6.3f}]")
    return results


if __name__ == "__main__":
    # Run the demonstration
    mle, X, y, true_theta = demonstrate_mle()
    
    # Visualize results (uncomment if you want to see plots)
    # visualize_mle_results(mle, X, y, true_theta)
    
    # Compare different noise levels
    compare_with_different_noise_levels()
    
    print("\n" + "=" * 60)
    print("DEMONSTRATION COMPLETE")
    print("=" * 60)
````

````python
import math
import random

class Matrix:
    """
    Simple matrix class for basic linear algebra operations
    """
    def __init__(self, data):
        if isinstance(data, (int, float)):
            # Scalar
            self.data = [[data]]
            self.rows = 1
            self.cols = 1
        elif isinstance(data[0], (int, float)):
            # Vector (convert to column matrix)
            self.data = [[x] for x in data]
            self.rows = len(data)
            self.cols = 1
        else:
            # Matrix
            self.data = [row[:] for row in data]  # Deep copy
            self.rows = len(data)
            self.cols = len(data[0]) if data else 0
    
    def __getitem__(self, key):
        if isinstance(key, tuple):
            i, j = key
            return self.data[i][j]
        else:
            return self.data[key]
    
    def __setitem__(self, key, value):
        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value
    
    def transpose(self):
        """Return transpose of matrix"""
        result = [[self.data[i][j] for i in range(self.rows)] 
                 for j in range(self.cols)]
        return Matrix(result)
    
    def multiply(self, other):
        """Matrix multiplication"""
        if isinstance(other, (int, float)):
            # Scalar multiplication
            result = [[self.data[i][j] * other for j in range(self.cols)] 
                     for i in range(self.rows)]
            return Matrix(result)
        
        # Matrix multiplication
        if self.cols != other.rows:
            raise ValueError(f"Cannot multiply {self.rows}x{self.cols} with {other.rows}x{other.cols}")
        
        result = [[0 for _ in range(other.cols)] for _ in range(self.rows)]
        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result[i][j] += self.data[i][j] * other.data[k][j]
        
        return Matrix(result)
    
    def add(self, other):
        """Matrix addition"""
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError("Matrices must have same dimensions for addition")
        
        result = [[self.data[i][j] + other.data[i][j] for j in range(self.cols)] 
                 for i in range(self.rows)]
        return Matrix(result)
    
    def subtract(self, other):
        """Matrix subtraction"""
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError("Matrices must have same dimensions for subtraction")
        
        result = [[self.data[i][j] - other.data[i][j] for j in range(self.cols)] 
                 for i in range(self.rows)]
        return Matrix(result)
    
    def to_vector(self):
        """Convert to flat list"""
        if self.cols == 1:
            return [self.data[i][0] for i in range(self.rows)]
        elif self.rows == 1:
            return self.data[0]
        else:
            raise ValueError("Cannot convert non-vector matrix to list")


def dot_product(a, b):
    """Compute dot product of two vectors"""
    if len(a) != len(b):
        raise ValueError("Vectors must have same length")
    return sum(a[i] * b[i] for i in range(len(a)))


def matrix_vector_multiply(matrix, vector):
    """Multiply matrix by vector"""
    if len(matrix[0]) != len(vector):
        raise ValueError("Matrix columns must equal vector length")
    
    result = []
    for i in range(len(matrix)):
        result.append(dot_product(matrix[i], vector))
    return result


def transpose_matrix(matrix):
    """Transpose a matrix"""
    return [[matrix[i][j] for i in range(len(matrix))] 
            for j in range(len(matrix[0]))]


def matrix_multiply(A, B):
    """Multiply two matrices"""
    if len(A[0]) != len(B):
        raise ValueError("Cannot multiply matrices with incompatible dimensions")
    
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result


def solve_linear_system(A, b):
    """
    Solve Ax = b using Gaussian elimination
    Simple implementation for 2x2 systems (can be extended)
    """
    n = len(A)
    
    # Create augmented matrix
    augmented = [A[i][:] + [b[i]] for i in range(n)]
    
    # Forward elimination
    for i in range(n):
        # Find pivot
        max_row = i
        for k in range(i + 1, n):
            if abs(augmented[k][i]) > abs(augmented[max_row][i]):
                max_row = k
        
        # Swap rows
        augmented[i], augmented[max_row] = augmented[max_row], augmented[i]
        
        # Make diagonal element 1
        pivot = augmented[i][i]
        if abs(pivot) < 1e-10:
            raise ValueError("Matrix is singular")
        
        for j in range(n + 1):
            augmented[i][j] /= pivot
        
        # Eliminate column
        for k in range(n):
            if k != i:
                factor = augmented[k][i]
                for j in range(n + 1):
                    augmented[k][j] -= factor * augmented[i][j]
    
    # Extract solution
    return [augmented[i][n] for i in range(n)]


class GaussianMLE:
    """
    Maximum Likelihood Estimation for Linear Regression with Gaussian Likelihood
    Pure Python implementation without external libraries
    """
    
    def __init__(self, sigma=1.0):
        """
        Initialize the MLE estimator
        
        Args:
            sigma (float): Standard deviation of Gaussian noise
        """
        self.sigma = sigma
        self.theta = None
        self.fitted = False
    
    def gaussian_pdf(self, x, mean=0, std=1):
        """
        Compute Gaussian probability density function
        PDF = (1/√(2π)σ) * exp(-0.5 * ((x-μ)/σ)²)
        """
        coefficient = 1.0 / (std * math.sqrt(2 * math.pi))
        exponent = -0.5 * ((x - mean) / std) ** 2
        return coefficient * math.exp(exponent)
    
    def gaussian_likelihood(self, y, x_theta):
        """
        Compute Gaussian likelihood p(y | x^T θ, σ²)
        
        Args:
            y (float): Observed label
            x_theta (float): Predicted value x^T θ
            
        Returns:
            float: Likelihood value
        """
        return self.gaussian_pdf(y, mean=x_theta, std=self.sigma)
    
    def log_likelihood_single(self, y, x_theta):
        """
        Compute log-likelihood for a single example
        log p(y | x^T θ, σ²) = -0.5 * log(2π) - log(σ) - 0.5 * ((y - x^T θ)/σ)²
        """
        diff = y - x_theta
        log_likelihood = -0.5 * math.log(2 * math.pi) - math.log(self.sigma) - 0.5 * (diff / self.sigma) ** 2
        return log_likelihood
    
    def negative_log_likelihood(self, theta, X, y):
        """
        Compute negative log-likelihood L(θ) = -∑ log p(y_n | x_n, θ)
        
        Args:
            theta (list): Parameter vector
            X (list of lists): Feature matrix (N x D)
            y (list): Label vector (N,)
            
        Returns:
            float: Negative log-likelihood value
        """
        total_nll = 0.0
        
        for i in range(len(y)):
            # Compute x_i^T θ
            x_theta = dot_product(X[i], theta)
            
            # Add log-likelihood for this example
            log_lik = self.log_likelihood_single(y[i], x_theta)
            total_nll -= log_lik  # Negative log-likelihood
        
        return total_nll
    
    def fit_analytical(self, X, y):
        """
        Fit using analytical solution: θ = (X^T X)^(-1) X^T y
        For Gaussian MLE, this is equivalent to least squares
        """
        # X^T (transpose)
        X_T = transpose_matrix(X)
        
        # X^T X
        XTX = matrix_multiply(X_T, X)
        
        # X^T y
        XTy = matrix_vector_multiply(X_T, y)
        
        # Solve (X^T X) θ = X^T y
        self.theta = solve_linear_system(XTX, XTy)
        self.fitted = True
    
    def fit_gradient_descent(self, X, y, learning_rate=0.01, max_iterations=1000, tolerance=1e-6):
        """
        Fit using gradient descent on negative log-likelihood
        """
        n_features = len(X[0])
        n_samples = len(X)
        
        # Initialize parameters randomly
        self.theta = [random.uniform(-1, 1) for _ in range(n_features)]
        
        for iteration in range(max_iterations):
            # Compute gradients
            gradients = [0.0] * n_features
            
            for i in range(n_samples):
                # Prediction for this sample
                prediction = dot_product(X[i], self.theta)
                error = y[i] - prediction
                
                # Gradient: ∂L/∂θ_j = -∑ (y_i - x_i^T θ) * x_ij / σ²
                for j in range(n_features):
                    gradients[j] -= error * X[i][j] / (self.sigma ** 2)
            
            # Update parameters
            old_theta = self.theta[:]
            for j in range(n_features):
                self.theta[j] -= learning_rate * gradients[j]
            
            # Check convergence
            change = sum((self.theta[j] - old_theta[j]) ** 2 for j in range(n_features))
            if math.sqrt(change) < tolerance:
                print(f"Converged after {iteration + 1} iterations")
                break
        
        self.fitted = True
    
    def fit(self, X, y, method='analytical'):
        """
        Fit the model using Maximum Likelihood Estimation
        
        Args:
            X (list of lists): Feature matrix (N x D)
            y (list): Label vector (N,)
            method (str): 'analytical' or 'gradient_descent'
        """
        # Add bias term (intercept) if not present
        if len(X[0]) == 1 or not all(abs(X[i][0] - 1.0) < 1e-10 for i in range(len(X))):
            X = [[1.0] + row for row in X]
        
        if method == 'analytical':
            self.fit_analytical(X, y)
        elif method == 'gradient_descent':
            self.fit_gradient_descent(X, y)
        else:
            raise ValueError("Method must be 'analytical' or 'gradient_descent'")
    
    def predict(self, X):
        """
        Make predictions using fitted model
        
        Args:
            X (list of lists): Feature matrix
            
        Returns:
            list: Predictions
        """
        if not self.fitted:
            raise ValueError("Model must be fitted before making predictions")
        
        # Add bias term if needed
        if len(X[0]) != len(self.theta):
            X = [[1.0] + row for row in X]
        
        predictions = []
        for i in range(len(X)):
            pred = dot_product(X[i], self.theta)
            predictions.append(pred)
        
        return predictions


def generate_synthetic_data(n_samples=100, true_theta=None, sigma=1.0, seed=42):
    """
    Generate synthetic data for testing MLE
    
    Args:
        n_samples (int): Number of samples
        true_theta (list): True parameter values [intercept, slope]
        sigma (float): Noise standard deviation
        seed (int): Random seed
        
    Returns:
        tuple: (X, y, true_theta)
    """
    random.seed(seed)
    
    if true_theta is None:
        true_theta = [2.0, -1.5]  # [intercept, slope]
    
    X = []
    y = []
    
    for i in range(n_samples):
        # Generate feature (uniform between -3 and 3)
        x_val = random.uniform(-3, 3)
        
        # Feature vector with bias term
        x_vec = [1.0, x_val]
        X.append(x_vec)
        
        # True label + Gaussian noise
        y_true = dot_product(x_vec, true_theta)
        noise = random.gauss(0, sigma)
        y.append(y_true + noise)
    
    return X, y, true_theta


def demonstrate_mle():
    """
    Demonstrate Maximum Likelihood Estimation with examples
    """
    print("=" * 60)
    print("MAXIMUM LIKELIHOOD ESTIMATION DEMONSTRATION")
    print("(Pure Python Implementation)")
    print("=" * 60)
    
    # Generate synthetic data
    print("\n1. Generating Synthetic Data")
    print("-" * 30)
    
    X, y, true_theta = generate_synthetic_data(n_samples=100, sigma=1.0)
    print(f"True parameters: θ = {true_theta}")
    print(f"Data shape: X = {len(X)}x{len(X[0])}, y = {len(y)}")
    print(f"First 5 samples:")
    for i in range(5):
        print(f"  x[{i}] = [{X[i][0]:.1f}, {X[i][1]:6.3f}], y[{i}] = {y[i]:6.3f}")
    
    # Fit MLE model
    print("\n2. Fitting MLE Model")
    print("-" * 30)
    
    # Analytical solution
    mle_analytical = GaussianMLE(sigma=1.0)
    mle_analytical.fit(X, y, method='analytical')
    print(f"MLE estimate (analytical): θ = [{mle_analytical.theta[0]:6.3f}, {mle_analytical.theta[1]:6.3f}]")
    
    # Gradient descent solution
    mle_gd = GaussianMLE(sigma=1.0)
    mle_gd.fit(X, y, method='gradient_descent')
    print(f"MLE estimate (grad. desc.): θ = [{mle_gd.theta[0]:6.3f}, {mle_gd.theta[1]:6.3f}]")
    
    # Compare with true values
    error_analytical = math.sqrt(sum((mle_analytical.theta[i] - true_theta[i])**2 for i in range(len(true_theta))))
    error_gd = math.sqrt(sum((mle_gd.theta[i] - true_theta[i])**2 for i in range(len(true_theta))))
    print(f"Error (analytical): {error_analytical:.4f}")
    print(f"Error (grad. desc.): {error_gd:.4f}")
    
    # Evaluate negative log-likelihood
    print("\n3. Negative Log-Likelihood Evaluation")
    print("-" * 30)
    
    nll_true = mle_analytical.negative_log_likelihood(true_theta, X, y)
    nll_estimated = mle_analytical.negative_log_likelihood(mle_analytical.theta, X, y)
    
    print(f"NLL at true parameters:      {nll_true:.4f}")
    print(f"NLL at estimated parameters: {nll_estimated:.4f}")
    print(f"Improvement: {nll_true - nll_estimated:.4f}")
    
    # Make predictions
    print("\n4. Making Predictions")
    print("-" * 30)
    
    X_test = [[x] for x in [-2, -1, 0, 1, 2]]  # Without bias term
    y_pred = mle_analytical.predict(X_test)
    
    print("Test predictions:")
    for i in range(len(X_test)):
        x_with_bias = [1.0] + X_test[i]
        y_true_test = dot_product(x_with_bias, true_theta)
        print(f"  x = {X_test[i][0]:5.2f}: pred = {y_pred[i]:6.3f}, true = {y_true_test:6.3f}")
    
    return mle_analytical, X, y, true_theta


def compare_with_different_noise_levels():
    """
    Compare MLE performance with different noise levels
    """
    print("\n5. Effect of Noise Level on MLE")
    print("-" * 30)
    
    sigma_values = [0.5, 1.0, 2.0, 3.0]
    true_theta = [2.0, -1.5]
    
    for sigma in sigma_values:
        # Generate data with different noise levels
        X, y, _ = generate_synthetic_data(n_samples=200, true_theta=true_theta, sigma=sigma)
        
        # Fit MLE
        mle = GaussianMLE(sigma=sigma)
        mle.fit(X, y, method='analytical')
        
        # Compute error
        error = math.sqrt(sum((mle.theta[i] - true_theta[i])**2 for i in range(len(true_theta))))
        
        print(f"σ = {sigma:.1f}: θ_estimated = [{mle.theta[0]:6.3f}, {mle.theta[1]:6.3f}], "
              f"error = {error:.4f}")
    
    print(f"\nTrue parameters: θ = [{true_theta[0]:6.3f}, {true_theta[1]:6.3f}]")


def demonstrate_likelihood_computation():
    """
    Demonstrate step-by-step likelihood computation
    """
    print("\n6. Step-by-Step Likelihood Computation")
    print("-" * 30)
    
    # Simple example with 3 data points
    X = [[1.0, 1.0], [1.0, 2.0], [1.0, 3.0]]
    y = [2.5, 1.0, -0.5]
    theta = [2.0, -1.0]
    
    mle = GaussianMLE(sigma=1.0)
    
    print("Data points:")
    for i in range(len(X)):
        print(f"  (x_{i+1}, y_{i+1}) = ({X[i][1]}, {y[i]})")
    
    print(f"\nParameter vector: θ = {theta}")
    print(f"Noise standard deviation: σ = {mle.sigma}")
    
    print("\nStep-by-step likelihood calculation:")
    total_log_likelihood = 0.0
    
    for i in range(len(X)):
        # Prediction
        prediction = dot_product(X[i], theta)
        
        # Individual likelihood
        likelihood = mle.gaussian_likelihood(y[i], prediction)
        log_likelihood = mle.log_likelihood_single(y[i], prediction)
        
        total_log_likelihood += log_likelihood
        
        print(f"  Point {i+1}: x^T θ = {prediction:.3f}, "
              f"p(y|x,θ) = {likelihood:.6f}, "
              f"log p(y|x,θ) = {log_likelihood:.6f}")
    
    negative_log_likelihood = -total_log_likelihood
    
    print(f"\nTotal log-likelihood: {total_log_likelihood:.6f}")
    print(f"Negative log-likelihood: {negative_log_likelihood:.6f}")
    
    # Verify with class method
    nll_class = mle.negative_log_likelihood(theta, X, y)
    print(f"Class method NLL: {nll_class:.6f}")
    print(f"Difference: {abs(negative_log_likelihood - nll_class):.8f}")


if __name__ == "__main__":
    # Run the demonstration
    mle, X, y, true_theta = demonstrate_mle()
    
    # Compare different noise levels
    compare_with_different_noise_levels()
    
    # Demonstrate likelihood computation
    demonstrate_likelihood_computation()
    
    print("\n" + "=" * 60)
    print("DEMONSTRATION COMPLETE")
    print("=" * 60)
````

## Example : Gaussian MLE Reduces to Least Squares

## Introduction

This notebook demonstrates a fundamental result in machine learning: **Maximum Likelihood Estimation with Gaussian likelihood is equivalent to least squares regression**. We'll work through the mathematical derivation step by step.



Fig.5 For the given data, the maximum likelihood estimate of the parameters results in the black diagonal line. The orange square shows the value of the maximum likelihood prediction at x = 60. Fig.6 Comparing the predictions with the maximum likelihood estimate and the MAP estimate at x = 60. The prior biases the slope to be less steep and the intercept to be closer to zero. In this example, the bias that moves the intercept closer to zero actually increases the slope.

## Starting Point: Gaussian Likelihood

From Example 8.4, we have the Gaussian likelihood for each data point:

$$p(y_n | x_n, \theta) = \mathcal{N}(y_n | x_n^T \theta, \sigma^2)$$

The negative log-likelihood for the entire dataset is:

$$L(\theta) = -\sum_{n=1}^N \log p(y_n | x_n, \theta)$$

## Step-by-Step Derivation

### Step 1: Substitute Gaussian Likelihood

Starting with the negative log-likelihood:

$$L(\theta) = -\sum_{n=1}^N \log p(y_n | x_n, \theta) = -\sum_{n=1}^N \log \mathcal{N}(y_n | x_n^T \theta, \sigma^2) \tag{8.18a}$$

### Step 2: Expand Gaussian Distribution

The Gaussian probability density function is:

$$\mathcal{N}(y | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y-\mu)^2}{2\sigma^2}\right)$$

Substituting this into our negative log-likelihood:

$$L(\theta) = -\sum_{n=1}^N \log \left[\frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y_n - x_n^T \theta)^2}{2\sigma^2}\right)\right] \tag{8.18b}$$

### Step 3: Apply Logarithm Properties

Using the logarithm properties $\log(ab) = \log(a) + \log(b)$ and $\log(\exp(x)) = x$:

$$L(\theta) = -\sum_{n=1}^N \left[\log\left(\frac{1}{\sqrt{2\pi\sigma^2}}\right) + \log\left(\exp\left(-\frac{(y_n - x_n^T \theta)^2}{2\sigma^2}\right)\right)\right]$$

$$L(\theta) = -\sum_{n=1}^N \left[-\log(\sqrt{2\pi\sigma^2}) - \frac{(y_n - x_n^T \theta)^2}{2\sigma^2}\right] \tag{8.18c}$$

### Step 4: Distribute the Negative Sign

$$L(\theta) = \sum_{n=1}^N \left[\log(\sqrt{2\pi\sigma^2}) + \frac{(y_n - x_n^T \theta)^2}{2\sigma^2}\right]$$

$$L(\theta) = \sum_{n=1}^N \log(\sqrt{2\pi\sigma^2}) + \sum_{n=1}^N \frac{(y_n - x_n^T \theta)^2}{2\sigma^2}$$

### Step 5: Separate Constant and Variable Terms

$$L(\theta) = \frac{1}{2\sigma^2} \sum_{n=1}^N (y_n - x_n^T \theta)^2 + \sum_{n=1}^N \log(\sqrt{2\pi\sigma^2}) \tag{8.18d}$$

## Key Insight: Connection to Least Squares

### Constant Term Analysis

The second term is **constant with respect to $\theta$**:

$$\sum_{n=1}^N \log(\sqrt{2\pi\sigma^2}) = N \log(\sqrt{2\pi\sigma^2})$$

Since $\sigma$ is given (fixed), this term doesn't affect the optimization.

### Variable Term Analysis

The first term is exactly the **least squares objective** (up to a constant factor):

$$\frac{1}{2\sigma^2} \sum_{n=1}^N (y_n - x_n^T \theta)^2$$

### Optimization Equivalence

Since we want to minimize $L(\theta)$ and the constant term doesn't depend on $\theta$:

$$\arg\min_{\theta} L(\theta) = \arg\min_{\theta} \left[\frac{1}{2\sigma^2} \sum_{n=1}^N (y_n - x_n^T \theta)^2\right]$$

The factor $\frac{1}{2\sigma^2}$ is positive and constant, so:

$$\arg\min_{\theta} L(\theta) = \arg\min_{\theta} \sum_{n=1}^N (y_n - x_n^T \theta)^2$$

**This is exactly the least squares problem from equation (8.8)!**

## Mathematical Summary

The complete derivation shows:

$$L(\theta) = \frac{1}{2\sigma^2} \underbrace{\sum_{n=1}^N (y_n - x_n^T \theta)^2}_{\text{Least Squares Term}} + \underbrace{N \log(\sqrt{2\pi\sigma^2})}_{\text{Constant Term}}$$

## Closed-Form Solution

Since Gaussian MLE reduces to least squares, we get the **closed-form solution**:

$$\hat{\theta}_{MLE} = (X^T X)^{-1} X^T y$$

where:
- $X$ is the $N \times D$ feature matrix
- $y$ is the $N \times 1$ label vector

## Practical Implications

### 1. **Computational Efficiency**
- No need for iterative optimization
- Direct matrix operations give exact solution
- Computationally efficient for moderate-sized problems

### 2. **Statistical Interpretation**
- Least squares has a probabilistic foundation
- Assumes Gaussian noise in observations
- MLE provides statistical properties (consistency, efficiency)

### 3. **Connection to Other Methods**
- Bridges frequentist (least squares) and Bayesian (MLE) approaches
- Foundation for understanding regularized regression
- Basis for generalized linear models

## Overfitting Considerations

Just like unregularized empirical risk minimization, **MLE can suffer from overfitting**:

- **High-dimensional data**: When $D \geq N$, $(X^T X)$ becomes singular
- **Noise sensitivity**: MLE can fit to noise in training data
- **Poor generalization**: May not perform well on test data

**Solution**: Use regularization techniques (covered in Section 8.3.2 on priors)

## Extension to Non-Gaussian Likelihoods

### When Gaussian Assumption Fails

For **non-Gaussian likelihood functions**:
- No closed-form solution exists
- Must use **numerical optimization methods** (Chapter 7)
- Examples: Logistic regression, Poisson regression

### Alternative Distributions
- **Laplace distribution**: Leads to L1 regularization (Lasso)
- **Student's t-distribution**: More robust to outliers
- **Exponential family**: Generalized linear models

## Visualization Interpretation

The figures mentioned in the text illustrate:

### Fig.5: MLE Regression Line
- **Black diagonal line**: Maximum likelihood estimate
- **Orange square**: MLE prediction at $x = 60$
- Shows how MLE fits a line through the data

### Fig.6: MLE vs MAP Comparison
- **MLE**: Unregularized solution (can overfit)
- **MAP**: Maximum A Posteriori (includes prior/regularization)
- **Effect of prior**: Biases slope to be less steep, intercept closer to zero

## Code Implementation Insight

The mathematical equivalence means:

```python

theta_mle = minimize_negative_log_likelihood(X, y, sigma)

# Method 2: Solve least squares directly  
theta_ls = solve((X.T @ X), X.T @ y)

```

## Summary

**Key Takeaway**: Maximum Likelihood Estimation with Gaussian likelihood is mathematically equivalent to least squares regression. This fundamental connection:

1. **Provides statistical foundation** for least squares
2. **Enables closed-form solutions** for linear regression
3. **Connects optimization and probability** theory
4. **Motivates regularization** techniques to prevent overfitting

This equivalence is one of the most important results in statistical learning theory, bridging classical regression with modern probabilistic machine learning.

````python
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Data from Table 8.2 ---
# Age (x_n) and Annual Salary (y_n in thousands)
data = [
    (36, 89.563),  # Aditya
    (47, 123.543),  # Bob
    (26, 23.989),  # Chloé
    (68, 138.769),  # Daisuke
    (33, 113.888),  # Elisabeth
]

# --- Data Preprocessing ---
def standardize(data_list):
    """Standardize a list to have mean 0 and variance 1."""
    n = len(data_list)
    mean = sum(data_list) / n
    variance = sum((x - mean) ** 2 for x in data_list) / n
    std_dev = math.sqrt(variance) if variance > 0 else 1
    return [(x - mean) / std_dev for x in data_list], mean, std_dev

# Extract x (age) and y (salary)
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]

# Standardize features
x_std, x_mean, x_std_dev = standardize(x_values)
y_std, y_mean, y_std_dev = standardize(y_values)

# Add bias term (x_n^{(0)} = 1) for affine function
X = [[1, x] for x in x_std]
y = y_std

# --- Matrix Operations ---
def dot_product(v1, v2):
    """Compute dot product of two vectors."""
    return sum(a * b for a, b in zip(v1, v2))

def matrix_multiply(A, B):
    """Multiply two matrices A (m x n) and B (n x p)."""
    m, n = len(A), len(A[0])
    n_b, p = len(B), len(B[0])
    result = [[0 for _ in range(p)] for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

def transpose_matrix(A):
    """Transpose a matrix."""
    m, n = len(A), len(A[0])
    return [[A[i][j] for i in range(m)] for j in range(n)]

def inverse_2x2(A):
    """Compute the inverse of a 2x2 matrix."""
    det = A[0][0] * A[1][1] - A[0][1] * A[1][0]
    if abs(det) < 1e-10:
        return [[0, 0], [0, 0]]  # Singular matrix approximation
    inv_det = 1 / det
    return [
        [A[1][1] * inv_det, -A[0][1] * inv_det],
        [-A[1][0] * inv_det, A[0][0] * inv_det]
    ]

# --- Gaussian MLE and Least Squares ---
def fit_gaussian_mle(X, y, sigma2=1.0):
    """
    Fit a linear model using Gaussian MLE by maximizing the log-likelihood.
    This reduces to solving the normal equations: (X^T X) theta = X^T y.
    Returns theta.
    """
    # The log-likelihood maximization leads to minimizing ||y - X theta||^2
    # Same as least squares, so we solve (X^T X) theta = X^T y
    X_T = transpose_matrix(X)
    XTX = matrix_multiply(X_T, X)
    XTy = matrix_multiply(X_T, [[yi] for yi in y])
    XTy = [row[0] for row in XTy]
    
    # Solve (X^T X) theta = X^T y
    XTX_inv = inverse_2x2(XTX)
    theta = matrix_multiply(XTX_inv, [[XTy[i]] for i in range(len(XTy))])
    theta = [row[0] for row in theta]
    
    return theta

def fit_least_squares(X, y):
    """
    Fit a linear model using least squares: min ||y - X theta||^2.
    Solves (X^T X) theta = X^T y.
    Returns theta.
    """
    # Identical to Gaussian MLE solution
    return fit_gaussian_mle(X, y)

def predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev):
    """Predict y for a given x, unstandardizing the result."""
    x_std = (x - x_mean) / x_std_dev
    x_input = [1, x_std]  # Add bias term
    y_std = dot_product(theta, x_input)
    y = y_std * y_std_dev + y_mean
    return y

# --- Compute Log-Likelihood and Empirical Risk ---
def compute_log_likelihood(X, y, theta, sigma2=1.0):
    """Compute the log-likelihood under Gaussian noise assumption."""
    N = len(y)
    residual_sum = sum((yi - dot_product(theta, xi)) ** 2 for xi, yi in zip(X, y))
    log_likelihood = - (N / 2) * math.log(2 * math.pi * sigma2) - (1 / (2 * sigma2)) * residual_sum
    return log_likelihood

def compute_empirical_risk(X, y, theta):
    """Compute the empirical risk (mean squared loss)."""
    N = len(y)
    predictions = [dot_product(theta, xi) for xi in X]
    return sum((yi - pred) ** 2 for yi, pred in zip(y, predictions)) / N

# --- Main Demonstration ---
def demonstrate_mle_least_squares():
    """Demonstrate the equivalence of Gaussian MLE and least squares."""
    print("=== Gaussian MLE and Least Squares Demonstration ===")
    print("Example 8.5: Gaussian MLE Reduces to Least Squares\n")

    # Step 1: Fit using Gaussian MLE
    print("Fitting Model Using Gaussian MLE")
    theta_mle = fit_gaussian_mle(X, y)
    print(f"MLE parameters: theta = {theta_mle}\n")
    
    # Step 2: Fit using Least Squares
    print("Fitting Model Using Least Squares")
    theta_ls = fit_least_squares(X, y)
    print(f"Least Squares parameters: theta = {theta_ls}\n")
    
    # Step 3: Verify equivalence
    print("Verifying Equivalence of MLE and Least Squares")
    equivalence = all(abs(mle - ls) < 1e-10 for mle, ls in zip(theta_mle, theta_ls))
    print(f"Parameters are equivalent: {equivalence}\n")
    
    # Step 4: Compute log-likelihood and empirical risk
    print("Computing Log-Likelihood and Empirical Risk")
    log_likelihood = compute_log_likelihood(X, y, theta_mle)
    emp_risk = compute_empirical_risk(X, y, theta_mle)
    print(f"Log-Likelihood: {log_likelihood:.4f}")
    print(f"Empirical Risk (Mean Squared Loss): {emp_risk:.4f}\n")
    
    # Step 5: Predict on new data
    print("Prediction on New Data")
    x_query = 60
    y_pred = predict(x_query, theta_mle, x_mean, x_std_dev, y_mean, y_std_dev)
    print(f"Prediction at x = {x_query}: y = {y_pred:.3f} thousand\n")
    
    return theta_mle, x_query, y_pred

# --- Plotting Function ---
def plot_results(theta, x_query, y_pred):
    """Plot the training data, fitted model, and prediction."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_values, y=y_values, color="blue", label="Training data")
    
    # Plot the linear fit
    x_range = [min(x_values) - 5, max(x_values) + 5]
    y_fit = [predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    sns.lineplot(x=x_range, y=y_fit, color="black", label="Fitted linear model")
    
    # Mark prediction at x = 60
    plt.scatter([x_query], [y_pred], color="red", s=100, marker="*", label=f"Prediction at x = {x_query}")
    plt.title("Gaussian MLE / Least-Squares Regression")
    plt.xlabel("x (age)")
    plt.ylabel("y (salary in thousands)")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Gaussian MLE and Least Squares Demonstration with Seaborn Plotting")
    print("=" * 60)

    # Demonstrate equivalence
    theta, x_query, y_pred = demonstrate_mle_least_squares()

    # Plot the results
    print("Plotting Training Data, Fitted Model, and Prediction")
    plot_results(theta, x_query, y_pred)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Fitted parameters: theta = {theta}")
    print(f"• Predicted salary at age {x_query}: {y_pred:.3f} thousand")
    print("• Plotted training data, fitted model, and prediction")
````

##  Maximum A Posteriori Estimation

If we have prior knowledge about the distribution of the parameters $\theta$, we can multiply an additional term to the likelihood. This additional term is a prior probability distribution on parameters $p(\theta)$. For a given prior, after observing some data $x$, how should we update the distribution of $\theta$? In other words, how should we represent the fact that we have more specific knowledge of $\theta$ after observing data $x$?

Bayes’ theorem, as discussed in Section 6.3, gives us a principled tool to update our probability distributions of random variables. It allows us to compute a *posterior* distribution $p(\theta | x)$ (the more specific knowledge) on the parameters $\theta$ from general prior statements (*prior* distribution) $p(\theta)$ and the function $p(x | \theta)$ that links the parameters \(\theta\) and the observed data \(x\) (called the *likelihood*):

$$
p(\theta | x) = \frac{p(x | \theta) p(\theta)}{p(x)}. \tag{8.19}
$$

Recall that we are interested in finding the parameter \(\theta\) that maximizes the posterior. Since the distribution \(p(x)\) does not depend on \(\theta\), we can ignore the value of the denominator for the optimization and obtain

$$
p(\theta | x) \propto p(x | \theta) p(\theta). \tag{8.20}
$$

The preceding proportion relation hides the density of the data $p(x)$, which may be difficult to estimate. Instead of estimating the minimum of the negative log-likelihood, we now estimate the minimum of the negative log-posterior, which is referred to as *maximum a posteriori estimation* (MAP estimation). An illustration of the effect of adding a zero-mean Gaussian prior is shown in Fig.6.

## Example 

In addition to the assumption of Gaussian likelihood in the previous example, we assume that the parameter vector is distributed as a multivariate Gaussian with zero mean, i.e., $p(\theta) = \mathcal{N}(0, \Sigma)$, where $\Sigma$ is the covariance matrix (Section 6.5). Note that the conjugate prior of a Gaussian is also a Gaussian (Section 6.6.1), and therefore we expect the posterior distribution to also be a Gaussian. We will see the details of maximum a posteriori estimation in Chapter 9.

The idea of including prior knowledge about where good parameters lie is widespread in machine learning. An alternative view, which we saw in Section 8.2.3, is the idea of *regularization*, which introduces an additional term that biases the resulting parameters to be close to the origin. Maximum a posteriori estimation can be considered to bridge the non-probabilistic and probabilistic worlds as it explicitly acknowledges the need for a prior distribution but it still only produces a point estimate of the parameters.

> **Remark**: The maximum likelihood estimate $\theta_{\text{ML}}$ possesses the following properties (Lehmann and Casella, 1998; Efron and Hastie, 2016):
> - **Asymptotic consistency**: The MLE converges to the true value in the limit of infinitely many observations, plus a random error that is approximately normal.
> - The size of the samples necessary to achieve these properties can be quite large.
> - The error’s variance decays in $1/N$, where $N$ is the number of data points.
> - Especially, in the “small” data regime, maximum likelihood estimation can lead to overfitting. $\diamond$

## Fig.7: Model Fitting

In a $\mathcal{M}_\theta$ parametrized class $\mathcal{M}_\theta$ of models, we optimize the model $\mathcal{M}_{\theta^*}$ parameters $\theta$ to minimize the distance to the true (unknown) model $\mathcal{M}^*$.

The principle of maximum likelihood estimation (and maximum a posteriori estimation) uses probabilistic modeling to reason about the uncertainty in the data and model parameters. However, we have not yet taken probabilistic modeling to its full extent. In this section, the resulting training procedure still produces a point estimate of the predictor, i.e., training returns one single set of parameter values that represent the best predictor. In Section 8.4, we will take the view that the parameter values should also be treated as random variables, and instead of estimating “best” values of that distribution, we will use the full parameter distribution when making predictions.



````python
import math
import matplotlib.pyplot as plt

# --- Data from Table 8.2 ---
# Age (x_n) and Annual Salary (y_n in thousands)
data = [
    (36, 89.563),  # Aditya
    (47, 123.543),  # Bob
    (26, 23.989),  # Chloé
    (68, 138.769),  # Daisuke
    (33, 113.888),  # Elisabeth
]

# --- Data Preprocessing ---
def standardize(data_list):
    """Standardize a list to have mean 0 and variance 1."""
    n = len(data_list)
    mean = sum(data_list) / n
    variance = sum((x - mean) ** 2 for x in data_list) / n
    std_dev = math.sqrt(variance) if variance > 0 else 1
    return [(x - mean) / std_dev for x in data_list], mean, std_dev

# Extract x (age) and y (salary)
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]

# Standardize features
x_std, x_mean, x_std_dev = standardize(x_values)
y_std, y_mean, y_std_dev = standardize(y_values)

# Add bias term (x_n^{(0)} = 1) for affine function
X = [[1, x] for x in x_std]
y = y_std

# --- Matrix Operations ---
def dot_product(v1, v2):
    """Compute dot product of two vectors."""
    return sum(a * b for a, b in zip(v1, v2))

def matrix_multiply(A, B):
    """Multiply two matrices A (m x n) and B (n x p)."""
    m, n = len(A), len(A[0])
    n_b, p = len(B), len(B[0])
    result = [[0 for _ in range(p)] for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

def transpose_matrix(A):
    """Transpose a matrix."""
    m, n = len(A), len(A[0])
    return [[A[i][j] for i in range(m)] for j in range(n)]

def inverse_2x2(A):
    """Compute the inverse of a 2x2 matrix."""
    det = A[0][0] * A[1][1] - A[0][1] * A[1][0]
    if abs(det) < 1e-10:
        return [[0, 0], [0, 0]]  # Singular matrix approximation
    inv_det = 1 / det
    return [
        [A[1][1] * inv_det, -A[0][1] * inv_det],
        [-A[1][0] * inv_det, A[0][0] * inv_det]
    ]

# --- MAP Estimation (Reduces to Regularized Least Squares) ---
def fit_map(X, y, lambda_reg):
    """
    Fit a linear model using MAP estimation with a Gaussian prior N(0, sigma_p^2 I).
    This reduces to regularized least squares: min (1/N) ||y - X theta||^2 + lambda ||theta||^2.
    lambda = sigma^2 / sigma_p^2, where sigma^2 is noise variance.
    Solves (X^T X + N*lambda I) theta = X^T y.
    Returns theta.
    """
    N = len(X)
    # Compute X^T X
    X_T = transpose_matrix(X)
    XTX = matrix_multiply(X_T, X)
    
    # Add regularization term from Gaussian prior: X^T X + N*lambda I
    # (Regularize only non-bias terms for simplicity)
    for i in range(1, len(XTX)):
        XTX[i][i] += N * lambda_reg
    
    # Compute X^T y
    XTy = matrix_multiply(X_T, [[yi] for yi in y])
    XTy = [row[0] for row in XTy]
    
    # Solve (X^T X + N*lambda I) theta = X^T y
    XTX_inv = inverse_2x2(XTX)
    theta = matrix_multiply(XTX_inv, [[XTy[i]] for i in range(len(XTy))])
    theta = [row[0] for row in theta]
    
    return theta

# --- MLE (Equivalent to Least Squares) ---
def fit_mle(X, y):
    """
    Fit a linear model using MLE (Gaussian likelihood), which reduces to least squares.
    Solves (X^T X) theta = X^T y.
    Returns theta.
    """
    X_T = transpose_matrix(X)
    XTX = matrix_multiply(X_T, X)
    XTy = matrix_multiply(X_T, [[yi] for yi in y])
    XTy = [row[0] for row in XTy]
    
    # Solve (X^T X) theta = X^T y
    XTX_inv = inverse_2x2(XTX)
    theta = matrix_multiply(XTX_inv, [[XTy[i]] for i in range(len(XTy))])
    theta = [row[0] for row in theta]
    
    return theta

def predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev):
    """Predict y for a given x, unstandardizing the result."""
    x_std = (x - x_mean) / x_std_dev
    x_input = [1, x_std]  # Add bias term
    y_std = dot_product(theta, x_input)
    y = y_std * y_std_dev + y_mean
    return y

# --- Compute Log-Posterior and Empirical Risk ---
def compute_log_posterior(X, y, theta, lambda_reg, sigma2=1.0):
    """Compute the log-posterior: log p(y|X,theta) + log p(theta)."""
    N = len(y)
    # Log-likelihood term
    residual_sum = sum((yi - dot_product(theta, xi)) ** 2 for xi, yi in zip(X, y))
    log_likelihood = - (N / 2) * math.log(2 * math.pi * sigma2) - (1 / (2 * sigma2)) * residual_sum
    # Log-prior term (Gaussian prior N(0, sigma_p^2 I), lambda = sigma^2 / sigma_p^2)
    theta_norm = sum(ti ** 2 for ti in theta[1:])  # Exclude bias term for prior
    sigma_p2 = sigma2 / lambda_reg if lambda_reg > 0 else 1e10  # Large sigma_p^2 if lambda=0
    log_prior = - (len(theta) - 1) / 2 * math.log(2 * math.pi * sigma_p2) - (1 / (2 * sigma_p2)) * theta_norm
    return log_likelihood + log_prior

def compute_empirical_risk(X, y, theta):
    """Compute the empirical risk (mean squared loss)."""
    N = len(y)
    predictions = [dot_product(theta, xi) for xi in X]
    return sum((yi - pred) ** 2 for yi, pred in zip(y, predictions)) / N

# --- Main Demonstration ---
def demonstrate_map_mle():
    """Demonstrate MAP estimation and compare with MLE."""
    print("=== MAP and MLE Demonstration ===")
    print("Section 8.3.2: Maximum A Posteriori Estimation\n")

    # Step 1: Fit using MLE (least squares)
    print("Fitting Model Using MLE (Least Squares)")
    theta_mle = fit_mle(X, y)
    print(f"MLE parameters: theta = {theta_mle}\n")
    
    # Step 2: Fit using MAP (regularized least squares)
    print("Fitting Model Using MAP (Regularized Least Squares)")
    lambda_reg = 0.1  # Regularization parameter (sigma^2 / sigma_p^2)
    theta_map = fit_map(X, y, lambda_reg)
    #print("MAP parameters (lambda = {}): theta = {}".format(lambda_reg, theta_map))
    print("\n")
    
    # Step 3: Compute log-posterior and empirical risk for MAP
    print("Computing Log-Posterior and Empirical Risk for MAP")
    log_posterior = compute_log_posterior(X, y, theta_map, lambda_reg)
    emp_risk_map = compute_empirical_risk(X, y, theta_map)
    print(f"Log-Posterior (MAP): {log_posterior:.4f}")
    print(f"Empirical Risk (MAP): {emp_risk_map:.4f}\n")
    
    # Step 4: Compute empirical risk for MLE
    print("Computing Empirical Risk for MLE")
    emp_risk_mle = compute_empirical_risk(X, y, theta_mle)
    print(f"Empirical Risk (MLE): {emp_risk_mle:.4f}\n")
    
    # Step 5: Predict on new data using MAP
    print("Prediction on New Data (Using MAP)")
    x_query = 60
    y_pred = predict(x_query, theta_map, x_mean, x_std_dev, y_mean, y_std_dev)
    print(f"Prediction at x = {x_query}: y = {y_pred:.3f} thousand\n")
    
    return theta_mle, theta_map, x_query, y_pred

# --- Plotting Function ---
def plot_results(theta_mle, theta_map, x_query, y_pred):
    """Plot the training data, fitted models (MLE and MAP), and prediction using Matplotlib."""
    plt.figure(figsize=(8, 6))
    
    # Plot training data
    plt.scatter(x_values, y_values, color="blue", label="Training data")
    
    # Plot MLE fit
    x_range = [min(x_values) - 5, max(x_values) + 5]
    y_fit_mle = [predict(x, theta_mle, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    plt.plot(x_range, y_fit_mle, color="green", label="MLE fit (Least Squares)")
    
    # Plot MAP fit
    y_fit_map = [predict(x, theta_map, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    plt.plot(x_range, y_fit_map, color="black", label="MAP fit (Regularized Least Squares)")
    
    # Mark prediction at x = 60
    plt.scatter([x_query], [y_pred], color="red", s=100, marker="*", label=f"Prediction at x = {x_query}")
    
    plt.title("MAP vs MLE: Linear Regression")
    plt.xlabel("x (age)")
    plt.ylabel("y (salary in thousands)")
    plt.legend()
    plt.grid(True)
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("MAP and MLE Demonstration with Matplotlib Plotting")
    print("=" * 60)

    # Demonstrate MAP and MLE
    theta_mle, theta_map, x_query, y_pred = demonstrate_map_mle()

    # Plot the results
    print("Plotting Training Data, Fitted Models, and Prediction")
    plot_results(theta_mle, theta_map, x_query, y_pred)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• MLE parameters: theta = {theta_mle}")
    print(f"• MAP parameters: theta = {theta_map}")
    print(f"• Predicted salary at age {x_query}: {y_pred:.3f} thousand")
    print("• Plotted training data, MLE fit, MAP fit, and prediction")
````
