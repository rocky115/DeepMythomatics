## Linear Regression

In the following, we will apply the mathematical concepts from Chapters 2, 5, 6, and 7 to solve linear regression (curve fitting) problems. In regression, we aim to find a function \( f \) that maps inputs \( \mathbf{x} \in \mathbb{R}^D \) to corresponding function values \( f(\mathbf{x}) \in \mathbb{R} \). We assume we are given a set of training inputs \( \mathbf{x}_n \) and corresponding noisy observations \( y_n = f(\mathbf{x}_n) + \epsilon \), where \( \epsilon \) is an i.i.d. random variable that describes measurement/observation noise and potentially unmodeled processes (which we will not consider further in this chapter). Throughout this chapter, we assume zero-mean Gaussian noise.

Our task is to find a function that not only models the training data but generalizes well to predicting function values at input locations that are not part of the training data (see Chapter 8). An illustration of such a regression problem is given in Figure 9.1.

A typical regression setting is given in Figure 9.1(a): For some input values $ \mathbf{x}_n $, we observe (noisy) function values $ y_n = f(\mathbf{x}_n) + \epsilon $. The task is to infer the function $ f $ that generated the data and generalizes well to function values at new input locations. A possible solution is given in Fig.1(b), where we also show three distributions centered at the function values $ f(\mathbf{x}) $ that represent the noise in the data.



### Fig.1: Illustration of Regression Problem

**Fig.1(a):** Regression problem: observed noisy function values from which we wish to infer the underlying function that generated the data.

**Fig.1(b):** Regression solution: possible function that could have generated the data (blue) with indication of the measurement noise of the function value at the corresponding inputs (orange distributions).

*(Note: The actual figures are not included here as they are illustrative. Figure 9.1(a) would show scattered data points, and Fig.1(b) would show a fitted curve with Gaussian distributions indicating noise at selected points.)*

Regression is a fundamental problem in machine learning, and regression problems appear in a diverse range of research areas and applications, including:

- **Time-series analysis** (e.g., system identification),
- **Control and robotics** (e.g., reinforcement learning, forward/inverse model learning),
- **Optimization** (e.g., line searches, global optimization),
- **Deep learning applications** (e.g., computer games, speech-to-text translation, image recognition, automatic video annotation).

Regression is also a key ingredient of classification algorithms. Finding a regression function requires solving a variety of problems, including the following:

1. **Choice of the model (type) and the parametrization of the regression function.** Given a dataset, what function classes (e.g., polynomials) are good candidates for modeling the data, and what particular parametrization (e.g., degree of the polynomial) should we choose? Model selection, as discussed in Section 8.6, allows us to compare various models to find the simplest model that explains the training data reasonably well.

2. **Finding good parameters.** Having chosen a model of the regression function, how do we find good model parameters? Here, we will need to look at different loss/objective functions (they determine what a "good" fit is) and optimization algorithms that allow us to minimize this loss.

3. **Overfitting and model selection.** Overfitting is a problem when the regression function fits the training data "too well" but does not generalize to unseen test data. Overfitting typically occurs if the underlying model (or its parametrization) is overly flexible and expressive; see Section 8.6. We will look at the underlying reasons and discuss ways to mitigate the effect of overfitting in the context of linear regression.

4. **Relationship between loss functions and parameter priors.** Loss functions (optimization objectives) are often motivated and induced by probabilistic models. We will look at the connection between loss functions and the underlying prior assumptions that induce these losses.

5. **Uncertainty modeling.** In any practical setting, we have access to only a finite, potentially large, amount of (training) data for selecting the model class and the corresponding parameters. Given that this finite amount of training data does not cover all possible scenarios, we may want to describe the remaining parameter uncertainty to obtain a measure of confidence of the model's prediction at test time; the smaller the training set, the more important uncertainty modeling. Consistent modeling of uncertainty equips model predictions with confidence bounds.

In the following, we will be using the mathematical tools from Chapters 3, 5, 6, and 7 to solve linear regression problems. We will discuss **maximum likelihood** and **maximum a posteriori (MAP)** estimation to find optimal model parameters. Using these parameter estimates, we will have a brief look at generalization errors and overfitting. Toward the end of this chapter, we will discuss **Bayesian linear regression**, which allows us to reason about model parameters at a higher level, thereby removing some of the problems encountered in maximum likelihood and MAP estimation.

```python
import math
import random
from typing import List, Dict, Tuple, Callable, Any
from abc import ABC, abstractmethod
from functools import reduce
import operator
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Helper function to compute binomial coefficient (n choose k)
def comb(n, k):
    if k < 0 or k > n:
        return 0
    k = min(k, n - k)
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

class BayesianModelSelection:
    def __init__(self, random_seed: int = 42):
        random.seed(random_seed)
        self.models = {}
        self.model_evidences = {}
        self.model_posteriors = {}
    
    def add_model(self, name: str, model: 'BayesianModel'):
        self.models[name] = model
    
    def compute_model_evidence(self, model_name: str, X: List[List[float]], 
                             y: List[float], method: str = 'analytical') -> float:
        if model_name not in self.models:
            raise ValueError(f"Model {model_name} not found")
        
        model = self.models[model_name]
        
        if method == 'analytical':
            return model.compute_analytical_evidence(X, y)
        elif method == 'laplace':
            return model.compute_laplace_evidence(X, y)
        elif method == 'sampling':
            return model.compute_sampling_evidence(X, y)
        else:
            raise ValueError(f"Unknown method: {method}")
    
    def compute_model_posteriors(self, X: List[List[float]], y: List[float], 
                               prior_probs: Dict[str, float] = None,
                               method: str = 'analytical') -> Dict[str, float]:
        if prior_probs is None:
            prior_probs = {name: 1.0/len(self.models) for name in self.models.keys()}
        
        log_evidences = {}
        for name in self.models.keys():
            log_evidences[name] = self.compute_model_evidence(name, X, y, method)
        
        log_posteriors = {}
        for name in self.models.keys():
            log_posteriors[name] = math.log(prior_probs[name]) + log_evidences[name]
        
        max_log_posterior = max(log_posteriors.values())
        exp_terms = {}
        for name in self.models.keys():
            exp_terms[name] = math.exp(log_posteriors[name] - max_log_posterior)
        
        normalization = sum(exp_terms.values())
        posteriors = {name: exp_terms[name] / normalization for name in self.models.keys()}
        
        self.model_evidences = {name: log_evidences[name] for name in self.models.keys()}
        self.model_posteriors = posteriors
        
        return posteriors
    
    def compute_bayes_factor(self, model1_name: str, model2_name: str, 
                           X: List[List[float]], y: List[float], 
                           method: str = 'analytical') -> Tuple[float, float]:
        if model1_name not in self.models or model2_name not in self.models:
            raise ValueError("One or both models not found")
        
        if not self.model_evidences or model1_name not in self.model_evidences or model2_name not in self.model_evidences:
            self.compute_model_posteriors(X, y, method=method)
        
        log_bf = self.model_evidences[model1_name] - self.model_evidences[model2_name]
        bf = math.exp(log_bf)
        return bf, log_bf
    
    def compute_posterior_odds(self, model1_name: str, model2_name: str, 
                             X: List[List[float]], y: List[float], 
                             prior_probs: Dict[str, float] = None,
                             method: str = 'analytical') -> Tuple[float, float]:
        if prior_probs is None:
            prior_probs = {name: 1.0/len(self.models) for name in self.models.keys()}
        
        self.compute_model_posteriors(X, y, prior_probs, method)
        posterior_odds = self.model_posteriors[model1_name] / self.model_posteriors[model2_name]
        log_posterior_odds = math.log(posterior_odds) if posterior_odds > 0 else float('-inf')
        return posterior_odds, log_posterior_odds
    
    def select_best_model(self, X: List[List[float]], y: List[float], 
                         prior_probs: Dict[str, float] = None,
                         method: str = 'analytical') -> Tuple[str, Dict[str, float]]:
        posteriors = self.compute_model_posteriors(X, y, prior_probs, method)
        best_model = max(posteriors.keys(), key=lambda k: posteriors[k])
        return best_model, posteriors
    
    def model_averaging_prediction(self, X_test: List[List[float]], 
                                 X_train: List[List[float]], y_train: List[float],
                                 prior_probs: Dict[str, float] = None) -> List[float]:
        posteriors = self.compute_model_posteriors(X_train, y_train, prior_probs)
        model_predictions = {}
        for name, model in self.models.items():
            model.fit(X_train, y_train)
            model_predictions[name] = model.predict(X_test)
        
        n_test = len(X_test)
        averaged_predictions = [0.0] * n_test
        for i in range(n_test):
            weighted_sum = 0.0
            for name in self.models.keys():
                weighted_sum += posteriors[name] * model_predictions[name][i]
            averaged_predictions[i] = weighted_sum
        return averaged_predictions
    
    def compute_bic(self, model_name: str, X: List[List[float]], y: List[float]) -> float:
        model = self.models[model_name]
        model.fit(X, y)
        log_likelihood = model.compute_log_likelihood(X, y)
        n_params = model.get_num_parameters()
        n_samples = len(X)
        bic = log_likelihood - (n_params / 2.0) * math.log(n_samples)
        return bic

class BayesianModel(ABC):
    @abstractmethod
    def fit(self, X: List[List[float]], y: List[float]):
        pass
    
    @abstractmethod
    def predict(self, X: List[List[float]]) -> List[float]:
        pass
    
    @abstractmethod
    def compute_analytical_evidence(self, X: List[List[float]], y: List[float]) -> float:
        pass
    
    @abstractmethod
    def compute_log_likelihood(self, X: List[List[float]], y: List[float]) -> float:
        pass
    
    @abstractmethod
    def get_num_parameters(self) -> int:
        pass
    
    def compute_laplace_evidence(self, X: List[List[float]], y: List[float]) -> float:
        self.fit(X, y)
        log_likelihood = self.compute_log_likelihood(X, y)
        hessian = self._compute_hessian(X, y)
        d = self.get_num_parameters()
        log_prior = self._compute_log_prior()
        log_det_hessian = self._log_determinant(hessian)
        laplace_evidence = (log_likelihood + log_prior + 
                          (d / 2.0) * math.log(2 * math.pi) - 
                          0.5 * log_det_hessian)
        return laplace_evidence
    
    def compute_sampling_evidence(self, X: List[List[float]], y: List[float], 
                                n_samples: int = 1000) -> float:
        log_weights = []
        for _ in range(n_samples):
            theta = self._sample_from_prior()
            self._set_parameters(theta)
            log_likelihood = self.compute_log_likelihood(X, y)
            log_weights.append(log_likelihood)
        max_log_weight = max(log_weights)
        log_evidence = max_log_weight + math.log(
            sum(math.exp(w - max_log_weight) for w in log_weights) / n_samples
        )
        return log_evidence
    
    def _compute_hessian(self, X: List[List[float]], y: List[float], eps: float = 1e-5):
        params = self._get_parameters()
        n_params = len(params)
        hessian = [[0.0] * n_params for _ in range(n_params)]
        for i in range(n_params):
            for j in range(n_params):
                if i == j:
                    params_plus = params[:]
                    params_plus[i] += eps
                    self._set_parameters(params_plus)
                    ll_plus = self.compute_log_likelihood(X, y)
                    params_minus = params[:]
                    params_minus[i] -= eps
                    self._set_parameters(params_minus)
                    ll_minus = self.compute_log_likelihood(X, y)
                    self._set_parameters(params)
                    ll_center = self.compute_log_likelihood(X, y)
                    hessian[i][j] = (ll_plus - 2*ll_center + ll_minus) / (eps**2)
                else:
                    params_pp = params[:]
                    params_pp[i] += eps
                    params_pp[j] += eps
                    self._set_parameters(params_pp)
                    ll_pp = self.compute_log_likelihood(X, y)
                    params_pm = params[:]
                    params_pm[i] += eps
                    params_pm[j] -= eps
                    self._set_parameters(params_pm)
                    ll_pm = self.compute_log_likelihood(X, y)
                    params_mp = params[:]
                    params_mp[i] -= eps
                    params_mp[j] += eps
                    self._set_parameters(params_mp)
                    ll_mp = self.compute_log_likelihood(X, y)
                    params_mm = params[:]
                    params_mm[i] -= eps
                    params_mm[j] -= eps
                    self._set_parameters(params_mm)
                    ll_mm = self.compute_log_likelihood(X, y)
                    hessian[i][j] = (ll_pp - ll_pm - ll_mp + ll_mm) / (4 * eps**2)
        self._set_parameters(params)
        return hessian
    
    def _log_determinant(self, matrix: List[List[float]]) -> float:
        n = len(matrix)
        lu = [row[:] for row in matrix]
        log_det = 0.0
        for i in range(n):
            max_row = i
            for k in range(i + 1, n):
                if abs(lu[k][i]) > abs(lu[max_row][i]):
                    max_row = k
            if max_row != i:
                lu[i], lu[max_row] = lu[max_row], lu[i]
                log_det -= math.log(abs(lu[i][i])) if lu[i][i] != 0 else float('-inf')
            else:
                log_det += math.log(abs(lu[i][i])) if lu[i][i] != 0 else float('-inf')
            for k in range(i + 1, n):
                if lu[i][i] != 0:
                    factor = lu[k][i] / lu[i][i]
                    for j in range(i, n):
                        lu[k][j] -= factor * lu[i][j]
        return log_det
    
    @abstractmethod
    def _get_parameters(self) -> List[float]:
        pass
    
    @abstractmethod
    def _set_parameters(self, params: List[float]):
        pass
    
    @abstractmethod
    def _sample_from_prior(self) -> List[float]:
        pass
    
    @abstractmethod
    def _compute_log_prior(self) -> float:
        pass

class BayesianLinearRegression(BayesianModel):
    def __init__(self, alpha: float = 1.0, beta: float = 1.0):
        self.alpha = alpha
        self.beta = beta
        self.weights = None
        self.weight_cov = None
        self.fitted = False
    
    def fit(self, X: List[List[float]], y: List[float]):
        X_bias = [[1.0] + x for x in X]
        n_samples = len(X_bias)
        n_features = len(X_bias[0])
        XTX = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                XTX[i][j] = sum(X_bias[k][i] * X_bias[k][j] for k in range(n_samples))
        S_N_inv = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                if i == j:
                    S_N_inv[i][j] = self.alpha + self.beta * XTX[i][j]
                else:
                    S_N_inv[i][j] = self.beta * XTX[i][j]
        self.weight_cov = self._matrix_inverse(S_N_inv)
        XTy = [0.0] * n_features
        for i in range(n_features):
            XTy[i] = sum(X_bias[k][i] * y[k] for k in range(n_samples))
        self.weights = [0.0] * n_features
        for i in range(n_features):
            self.weights[i] = self.beta * sum(self.weight_cov[i][j] * XTy[j] 
                                            for j in range(n_features))
        self.fitted = True
    
    def predict(self, X: List[List[float]]) -> List[float]:
        if not self.fitted:
            raise ValueError("Model must be fitted before prediction")
        X_bias = [[1.0] + x for x in X]
        predictions = []
        for x in X_bias:
            pred = sum(w * f for w, f in zip(self.weights, x))
            predictions.append(pred)
        return predictions
    
    def compute_analytical_evidence(self, X: List[List[float]], y: List[float]) -> float:
        self.fit(X, y)
        X_bias = [[1.0] + x for x in X]
        n_samples = len(X_bias)
        n_features = len(X_bias[0])
        log_evidence = 0.0
        log_evidence += (n_features / 2.0) * math.log(self.alpha)
        log_evidence += (n_samples / 2.0) * math.log(self.beta)
        log_evidence -= (n_samples / 2.0) * math.log(2 * math.pi)
        log_det_S_N = self._log_determinant(self.weight_cov)
        log_evidence += 0.5 * log_det_S_N
        y_squared = sum(yi**2 for yi in y)
        log_evidence -= (self.beta / 2.0) * y_squared
        XTy = [0.0] * n_features
        for i in range(n_features):
            XTy[i] = sum(X_bias[k][i] * y[k] for k in range(n_samples))
        mN_XTy = sum(self.weights[i] * XTy[i] for i in range(n_features))
        log_evidence += (self.beta / 2.0) * mN_XTy
        return log_evidence
    
    def compute_log_likelihood(self, X: List[List[float]], y: List[float]) -> float:
        if not self.fitted:
            self.fit(X, y)
        predictions = self.predict(X)
        n_samples = len(y)
        log_likelihood = -(n_samples / 2.0) * math.log(2 * math.pi / self.beta)
        residual_sum = sum((yi - pred)**2 for yi, pred in zip(y, predictions))
        log_likelihood -= (self.beta / 2.0) * residual_sum
        return log_likelihood
    
    def get_num_parameters(self) -> int:
        if self.weights is None:
            return 0
        return len(self.weights)
    
    def _get_parameters(self) -> List[float]:
        return self.weights[:] if self.weights else []
    
    def _set_parameters(self, params: List[float]):
        self.weights = params[:]
    
    def _sample_from_prior(self) -> List[float]:
        if self.weights is None:
            return []
        n_params = len(self.weights)
        std = math.sqrt(1.0 / self.alpha)
        return [random.gauss(0, std) for _ in range(n_params)]
    
    def _compute_log_prior(self) -> float:
        if self.weights is None:
            return 0.0
        n_params = len(self.weights)
        log_prior = (n_params / 2.0) * math.log(self.alpha / (2 * math.pi))
        weight_sum_squares = sum(w**2 for w in self.weights)
        log_prior -= (self.alpha / 2.0) * weight_sum_squares
        return log_prior
    
    def _matrix_inverse(self, matrix: List[List[float]]) -> List[List[float]]:
        n = len(matrix)
        augmented = []
        for i in range(n):
            row = matrix[i][:] + [0.0] * n
            row[n + i] = 1.0
            augmented.append(row)
        for i in range(n):
            max_row = i
            for k in range(i + 1, n):
                if abs(augmented[k][i]) > abs(augmented[max_row][i]):
                    max_row = k
            augmented[i], augmented[max_row] = augmented[max_row], augmented[i]
            pivot = augmented[i][i]
            if abs(pivot) < 1e-10:
                raise ValueError("Matrix is singular")
            for j in range(2 * n):
                augmented[i][j] /= pivot
            for k in range(n):
                if k != i:
                    factor = augmented[k][i]
                    for j in range(2 * n):
                        augmented[k][j] -= factor * augmented[i][j]
        inverse = []
        for i in range(n):
            inverse.append(augmented[i][n:])
        return inverse

class BayesianPolynomialRegression(BayesianLinearRegression):
    def __init__(self, degree: int, alpha: float = 1.0, beta: float = 1.0):
        super().__init__(alpha, beta)
        self.degree = degree
    
    def _transform_features(self, X: List[List[float]]) -> List[List[float]]:
        X_poly = []
        for x in X:
            x_val = x[0] if x else 0.0
            poly_features = []
            for d in range(self.degree + 1):
                poly_features.append(x_val ** d)
            X_poly.append(poly_features)
        return X_poly
    
    def fit(self, X: List[List[float]], y: List[float]):
        X_poly = self._transform_features(X)
        super().fit(X_poly, y)
    
    def predict(self, X: List[List[float]]) -> List[float]:
        X_poly = self._transform_features(X)
        return super().predict(X_poly)
    
    def compute_analytical_evidence(self, X: List[List[float]], y: List[float]) -> float:
        X_poly = self._transform_features(X)
        return super().compute_analytical_evidence(X_poly, y)
    
    def compute_log_likelihood(self, X: List[List[float]], y: List[float]) -> float:
        X_poly = self._transform_features(X)
        return super().compute_log_likelihood(X_poly, y)

def generate_polynomial_data(n_samples: int = 50, noise_std: float = 0.1, 
                           true_degree: int = 2) -> Tuple[List[List[float]], List[float]]:
    random.seed(42)
    X = []
    y = []
    for _ in range(n_samples):
        x_val = random.uniform(-1, 1)
        if true_degree == 1:
            y_val = 0.5 + 1.2 * x_val
        elif true_degree == 2:
            y_val = 0.5 + 1.2 * x_val - 0.8 * x_val**2
        elif true_degree == 3:
            y_val = 0.5 + 1.2 * x_val - 0.8 * x_val**2 + 0.3 * x_val**3
        else:
            y_val = 0.5 + 1.2 * x_val
        y_val += random.gauss(0, noise_std)
        X.append([x_val])
        y.append(y_val)
    return X, y

def main():
    # Set Seaborn style
    sns.set(style="whitegrid")
    
    print("="*60)
    print("BAYESIAN MODEL SELECTION AND BAYES FACTORS DEMONSTRATION")
    print("="*60)
    
    # Generate synthetic data
    print("\n1. Generating synthetic polynomial data...")
    X, y = generate_polynomial_data(n_samples=50, noise_std=0.15, true_degree=2)
    print(f"   Dataset: {len(X)} samples")
    print(f"   Feature range: [{min(x[0] for x in X):.3f}, {max(x[0] for x in X):.3f}]")
    print(f"   Target range: [{min(y):.3f}, {max(y):.3f}]")
    
    # Initialize Bayesian Model Selection
    bms = BayesianModelSelection(random_seed=42)
    
    # Add candidate models
    print("\n2. Adding candidate models...")
    models_config = {
        'Linear': BayesianPolynomialRegression(degree=1, alpha=1.0, beta=25.0),
        'Quadratic': BayesianPolynomialRegression(degree=2, alpha=1.0, beta=25.0),
        'Cubic': BayesianPolynomialRegression(degree=3, alpha=1.0, beta=25.0),
        'Quartic': BayesianPolynomialRegression(degree=4, alpha=1.0, beta=25.0)
    }
    
    for name, model in models_config.items():
        bms.add_model(name, model)
        print(f"   Added {name} model (degree {model.degree})")
    
    # Compute model posteriors with uniform priors
    print("\n3. Computing model posteriors (uniform priors)...")
    posteriors = bms.compute_model_posteriors(X, y, method='analytical')
    print(f"   Model Evidence (log):")
    for name in models_config.keys():
        evidence = bms.model_evidences[name]
        print(f"   {name:12s}: {evidence:8.3f}")
    print(f"\n   Model Posteriors:")
    for name, posterior in posteriors.items():
        print(f"   {name:12s}: {posterior:8.4f} ({posterior*100:5.1f}%)")
    
    # Plot 1: Model Posteriors
    print("\n   Plotting model posteriors...")
    plt.figure(figsize=(8, 6))
    sns.barplot(x=list(posteriors.values()), y=list(posteriors.keys()))
    plt.title("Posterior Probabilities of Models")
    plt.xlabel("Posterior Probability")
    plt.ylabel("Model")
    plt.show()
    
    # Compute Bayes Factors
    print("\n4. Computing Bayes Factors...")
    model_pairs = [
        ('Quadratic', 'Linear'),
        ('Quadratic', 'Cubic'),
        ('Quadratic', 'Quartic'),
        ('Cubic', 'Linear'),
        ('Quartic', 'Cubic')
    ]
    
    bayes_factors = {}
    log_bayes_factors = {}
    pair_labels = []
    print(f"   {'Model 1':12s} vs {'Model 2':12s}: {'Bayes Factor':>12s} {'log(BF)':>8s}")
    print(f"   {'-'*12}    {'-'*12}    {'-'*12} {'-'*8}")
    for m1, m2 in model_pairs:
        bf, log_bf = bms.compute_bayes_factor(m1, m2, X, y, method='analytical')
        bayes_factors[f"{m1} vs {m2}"] = bf
        log_bayes_factors[f"{m1} vs {m2}"] = log_bf
        pair_labels.append(f"{m1} vs {m2}")
        print(f"   {m1:12s} vs {m2:12s}: {bf:12.3f} {log_bf:8.3f}")
    
    # Plot 2: Log Bayes Factors
    print("\n   Plotting log Bayes Factors...")
    plt.figure(figsize=(8, 6))
    sns.barplot(x=list(log_bayes_factors.values()), y=pair_labels)
    plt.title("Log Bayes Factors Between Models")
    plt.xlabel("Log Bayes Factor")
    plt.ylabel("Model Pair")
    plt.axvline(x=0, color='black', linestyle='--')
    plt.show()
    
    # Demonstrate effect of prior odds
    print("\n5. Effect of Prior Odds on Posterior Odds...")
    prior_probs = {
        'Linear': 0.4,
        'Quadratic': 0.3,
        'Cubic': 0.2,
        'Quartic': 0.1
    }
    print(f"   Custom Prior Probabilities:")
    for name, prior in prior_probs.items():
        print(f"   {name:12s}: {prior:8.4f}")
    
    print(f"\n   {'Model 1':12s} vs {'Model 2':12s}: {'Prior Odds':>12s} {'Bayes Factor':>12s} {'Posterior Odds':>14s}")
    print(f"   {'-'*12}    {'-'*12}    {'-'*12} {'-'*12} {'-'*14}")
    for m1, m2 in model_pairs:
        bf, _ = bms.compute_bayes_factor(m1, m2, X, y, method='analytical')
        prior_odds = prior_probs[m1] / prior_probs[m2]
        posterior_odds, _ = bms.compute_posterior_odds(m1, m2, X, y, prior_probs, method='analytical')
        print(f"   {m1:12s} vs {m2:12s}: {prior_odds:12.3f} {bf:12.3f} {posterior_odds:14.3f}")
    
    # Demonstrate Jeffreys-Lindley paradox with diffuse prior
    print("\n6. Jeffreys-Lindley Paradox (Diffuse Prior Effect)...")
    bms_diffuse = BayesianModelSelection(random_seed=42)
    models_config_diffuse = {
        'Linear': BayesianPolynomialRegression(degree=1, alpha=1.0, beta=25.0),
        'Quadratic': BayesianPolynomialRegression(degree=2, alpha=1.0, beta=25.0),
        'Quartic_Diffuse': BayesianPolynomialRegression(degree=4, alpha=0.01, beta=25.0)
    }
    
    for name, model in models_config_diffuse.items():
        bms_diffuse.add_model(name, model)
    
    posteriors_diffuse = bms_diffuse.compute_model_posteriors(X, y, method='analytical')
    print(f"   Model Evidence with Diffuse Prior (log):")
    for name in models_config_diffuse.keys():
        evidence = bms_diffuse.model_evidences[name]
        print(f"   {name:12s}: {evidence:8.3f}")
    print(f"\n   Model Posteriors with Diffuse Prior:")
    for name, posterior in posteriors_diffuse.items():
        print(f"   {name:12s}: {posterior:8.4f} ({posterior*100:5.1f}%)")
    
    # Plot 3: Compare Posteriors (Standard vs Diffuse Priors)
    print("\n   Plotting standard vs diffuse prior posteriors...")
    plt.figure(figsize=(8, 6))
    # Prepare data for plotting
    models_standard = list(posteriors.keys())
    models_diffuse = list(posteriors_diffuse.keys())
    posteriors_standard = list(posteriors.values())
    posteriors_diffuse_list = [posteriors_diffuse.get(model, 0.0) for model in models_standard]
    bar_width = 0.35
    x = np.arange(len(models_standard))
    plt.bar(x - bar_width/2, posteriors_standard, bar_width, label='Standard Prior', color='skyblue')
    plt.bar(x + bar_width/2, posteriors_diffuse_list, bar_width, label='Diffuse Prior (Quartic)', color='salmon')
    plt.xlabel('Model')
    plt.ylabel('Posterior Probability')
    plt.title('Effect of Diffuse Prior on Model Posteriors')
    plt.xticks(x, models_standard)
    plt.legend()
    plt.show()
    
    # Model averaging predictions
    print("\n7. Demonstrating Bayesian Model Averaging...")
    train_size = int(0.8 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    print(f"   Training samples: {len(X_train)}")
    print(f"   Test samples: {len(X_test)}")
    
    averaged_predictions = bms.model_averaging_prediction(X_test, X_train, y_train)
    individual_predictions = {}
    for name, model in bms.models.items():
        model.fit(X_train, y_train)
        individual_predictions[name] = model.predict(X_test)
    
    def mean_squared_error(y_true, y_pred):
        return sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred)) / len(y_true)
    
    print("\n   Mean Squared Error on Test Set:")
    print(f"   {'Model':12s}: {'MSE':>8s}")
    print(f"   {'-'*12}: {'-'*8}")
    for name, preds in individual_predictions.items():
        mse = mean_squared_error(y_test, preds)
        print(f"   {name:12s}: {mse:8.4f}")
    mse_averaged = mean_squared_error(y_test, averaged_predictions)
    print(f"   {'Averaged':12s}: {mse_averaged:8.4f}")
    
    # Plot 4: Data and Model Fits
    print("\n   Plotting data and model fits...")
    # Prepare data for plotting
    X_full = X_train + X_test
    y_full = y_train + y_test
    X_vals = [x[0] for x in X_full]
    # Sort data for smooth plotting
    sorted_indices = np.argsort(X_vals)
    X_sorted = [X_vals[i] for i in sorted_indices]
    y_sorted = [y_full[i] for i in sorted_indices]
    X_sorted_array = [[x] for x in X_sorted]
    
    plt.figure(figsize=(10, 6))
    # Plot data points
    sns.scatterplot(x=X_vals, y=y_full, color='black', label='Data', alpha=0.5)
    
    # Plot model fits
    colors = {'Linear': 'blue', 'Quadratic': 'green', 'Cubic': 'orange', 'Quartic': 'red', 'Averaged': 'purple'}
    for name, model in bms.models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_sorted_array)
        sns.lineplot(x=X_sorted, y=preds, label=name, color=colors[name])
    
    # Plot averaged predictions
    averaged_preds = bms.model_averaging_prediction(X_sorted_array, X_train, y_train)
    sns.lineplot(x=X_sorted, y=averaged_preds, label='Averaged', color=colors['Averaged'], linewidth=2, linestyle='--')
    
    plt.title("Data and Model Fits")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
