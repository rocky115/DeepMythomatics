#  Empirical Risk Minimization

After having all the mathematics under our belt, we are now in a position to introduce what it means to learn. The “learning” part of machine learning boils down to estimating parameters based on training data. In this section, we consider the case of a predictor that is a function, and consider the case of probabilistic models in Section 8.3. We describe the idea of *empirical risk minimization*, which was originally popularized by the proposal of the support vector machine (described in Chapter 12). However, its general principles are widely applicable and allow us to ask the question of what is learning without explicitly constructing probabilistic models.

There are four main design choices, which we will cover in detail in the following subsections:

- **Section 8.2.1**: What is the set of functions we allow the predictor to take?
- **Section 8.2.2**: How do we measure how well the predictor performs on the training data?
- **Section 8.2.3**: How do we construct predictors from only training data that performs well on unseen test data?
- **Section 8.2.4**: What is the procedure for searching over the space of models?

## Hypothesis Class of Functions

Assume we are given $N$ examples $x_n \in \mathbb{R}^D$ and corresponding scalar labels $y_n \in \mathbb{R}$. We consider the supervised learning setting, where we obtain pairs $(x_1, y_1), \ldots, (x_N, y_N)$. Given this data, we would like to estimate a predictor $f(\cdot, \theta): \mathbb{R}^D \to \mathbb{R}$, parametrized by $\theta$. We hope to be able to find a good parameter $\theta^*$ such that we fit the data well, that is,

$$
f(x_n, \theta^*) \approx y_n \quad \text{for all } n = 1, \ldots, N. \tag{8.3}
$$

In this section, we use the notation $\hat{y}_n = f(x_n, \theta^*)$ to represent the output of the predictor.

> **Remark**: For ease of presentation, we will describe empirical risk minimization in terms of supervised learning (where we have labels). This simplifies the definition of the hypothesis class and the loss function. It is also common in machine learning to choose a parametrized class of functions, for example affine functions. $\diamond$

## Example .1

We introduce the problem of ordinary least-squares regression to illustrate empirical risk minimization. A more comprehensive account of regression is given in Chapter 9. When the label $y_n$ is real-valued, a popular choice of function class for predictors is the set of affine functions. We choose a more compact notation for an affine function by concatenating an additional unit feature $x_n^{(0)} = 1$ to $x_n$, i.e., $x_n = [1, x_n^{(1)}, x_n^{(2)}, \ldots, x_n^{(D)}]^\top$. The parameter vector is correspondingly $\theta = [\theta_0, \theta_1, \theta_2, \ldots, \theta_D]^\top$, allowing us to write the predictor as a linear function

$$
f(x_n, \theta) = \theta^\top x_n. \tag{8.4}
$$

This linear predictor is equivalent to the affine model

$$
f(x_n, \theta) = \theta_0 + \sum_{d=1}^D \theta_d x_n^{(d)}. \tag{8.5}
$$

The predictor takes the vector of features representing a single example $x_n$ as input and produces a real-valued output, i.e., $f: \mathbb{R}^{D+1} \to \mathbb{R}$. The

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
from random import shuffle

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Data from Table 8.2 ---
# Age (x_n) and Annual Salary (y_n in thousands)
data = [
    (36, 89.563),  # Aditya
    (47, 123.543),  # Bob
    (26, 23.989),  # Chloé
    (68, 138.769),  # Daisuke
    (33, 113.888),  # Elisabeth
]

# --- Data Preprocessing ---
def standardize(data_list):
    """Standardize a list to have mean 0 and variance 1."""
    n = len(data_list)
    mean = sum(data_list) / n
    variance = sum((x - mean) ** 2 for x in data_list) / n
    std_dev = math.sqrt(variance) if variance > 0 else 1
    return [(x - mean) / std_dev for x in data_list], mean, std_dev

# Extract x (age) and y (salary)
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]

# Standardize features
x_std, x_mean, x_std_dev = standardize(x_values)
y_std, y_mean, y_std_dev = standardize(y_values)

# Add bias term (x_n^{(0)} = 1) for affine function
x_std_with_bias = [[1, x] for x in x_std]
data_with_bias = list(zip(x_std_with_bias, y_std))

# --- Empirical Risk Minimization (Least Squares with Regularization) ---
def dot_product(v1, v2):
    """Compute dot product of two vectors."""
    return sum(a * b for a, b in zip(v1, v2))

def fit_linear_model(X, y, lambda_reg=0):
    """
    Fit a linear model f(x_n, theta) = theta^T x_n using least squares with L2 regularization.
    X: list of [1, x] (with bias term), y: list of labels.
    Returns theta (including theta_0).
    """
    N = len(X)
    D = len(X[0])  # D+1 dimensions (including bias)
    
    # Compute X^T X + lambda * I (excluding bias term for regularization)
    XTX = [[0] * D for _ in range(D)]
    for n in range(N):
        for i in range(D):
            for j in range(D):
                XTX[i][j] += X[n][i] * X[n][j]
    for i in range(1, D):  # Regularize only non-bias terms
        XTX[i][i] += lambda_reg * N
    
    # Compute X^T y
    XTy = [0] * D
    for n in range(N):
        for i in range(D):
            XTy[i] += X[n][i] * y[n]
    
    # Solve (X^T X + lambda I) theta = X^T y using Gaussian elimination
    # (Simplified for 2D: bias + 1 feature)
    A = XTX
    b = XTy
    theta = [0] * D
    for i in range(D):
        for j in range(i + 1, D):
            factor = A[j][i] / A[i][i] if A[i][i] != 0 else 0
            for k in range(D):
                A[j][k] -= factor * A[i][k]
            b[j] -= factor * b[i]
    for i in range(D - 1, -1, -1):
        theta[i] = b[i]
        for j in range(i + 1, D):
            theta[i] -= A[i][j] * theta[j]
        theta[i] /= A[i][i] if A[i][i] != 0 else 1
    
    return theta

def predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev):
    """Predict y for a given x, unstandardizing the result."""
    x_std = (x - x_mean) / x_std_dev
    x_input = [1, x_std]  # Add bias term
    y_std = dot_product(theta, x_input)
    y = y_std * y_std_dev + y_mean
    return y

# --- Cross-Validation ---
def mean_squared_error(y_true, y_pred):
    """Compute mean squared error."""
    n = len(y_true)
    return sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred)) / n

def k_fold_cross_validation(data, k, lambda_reg):
    """
    Perform k-fold cross-validation to evaluate the model.
    Returns average validation MSE.
    """
    shuffle(data)  # Randomly shuffle data
    fold_size = len(data) // k
    mse_scores = []
    
    for i in range(k):
        # Split into training and validation sets
        val_start = i * fold_size
        val_end = (i + 1) * fold_size if i < k - 1 else len(data)
        val_data = data[val_start:val_end]
        train_data = data[:val_start] + data[val_end:]
        
        # Extract X and y
        X_train = [d[0] for d in train_data]
        y_train = [d[1] for d in train_data]
        X_val = [d[0] for d in val_data]
        y_val = [d[1] for d in val_data]
        
        # Fit model on training data
        theta = fit_linear_model(X_train, y_train, lambda_reg)
        
        # Predict on validation data
        y_pred = [dot_product(theta, x) for x in X_val]
        
        # Compute MSE
        mse = mean_squared_error(y_val, y_pred)
        mse_scores.append(mse)
    
    return sum(mse_scores) / len(mse_scores)

# --- Main Demonstration ---
def demonstrate_erm():
    """Demonstrate empirical risk minimization for linear regression."""
    print("=== Empirical Risk Minimization Demonstration ===")
    print("Section 8.2: Empirical Risk Minimization\n")

    # Step 1: Cross-validation to evaluate generalization
    print("Cross-Validation to Evaluate Generalization (Section 8.2.4)")
    k = 3
    lambda_reg = 0.1  # Regularization parameter (Section 8.2.3)
    avg_mse = k_fold_cross_validation(data_with_bias, k, lambda_reg)
    print(f"Average MSE from {k}-fold cross-validation: {avg_mse:.4f}\n")
    
    # Step 2: Fit the model on all training data
    print("Fitting the Model (Section 8.2.1: Hypothesis Class)")
    theta = fit_linear_model(x_std_with_bias, y_std, lambda_reg)
    print(f"Fitted parameters: theta = {theta}\n")
    
    # Step 3: Predict on new data
    print("Prediction on New Data")
    x_query = 60
    y_pred = predict(x_query, theta, x_mean, x_std_dev, y_mean, y_std_dev)
    print(f"Prediction at x = {x_query}: y = {y_pred:.3f} thousand\n")
    
    return theta, x_query, y_pred

# --- Plotting Function ---
def plot_results(theta, x_query, y_pred):
    """Plot the training data, fitted model, and prediction."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_values, y=y_values, color="blue", label="Training data")
    
    # Plot the linear fit
    x_range = [min(x_values) - 5, max(x_values) + 5]
    y_fit = [predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    sns.lineplot(x=x_range, y=y_fit, color="black", label="Fitted linear model")
    
    # Mark prediction at x = 60
    plt.scatter([x_query], [y_pred], color="red", s=100, marker="*", label=f"Prediction at x = {x_query}")
    plt.title("Empirical Risk Minimization: Linear Regression")
    plt.xlabel("x (age)")
    plt.ylabel("y (salary in thousands)")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Empirical Risk Minimization Demonstration with Seaborn Plotting")
    print("=" * 60)

    # Demonstrate ERM
    theta, x_query, y_pred = demonstrate_erm()

    # Plot the results
    print("Plotting Training Data, Fitted Model, and Prediction")
    plot_results(theta, x_query, y_pred)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Fitted model parameters: theta = {theta}")
    print(f"• Predicted salary at age {x_query}: {y_pred:.3f} thousand")
    print("• Plotted training data, fitted model, and prediction")
```

## 8.2.2 Loss Function for Training

Consider the label $y_n$ for a particular example; and the corresponding prediction $\hat{y}_n$ that we make based on $x_n$. To define what it means to fit the data well, we need to specify a *loss function* $\ell(y_n, \hat{y}_n)$ that takes the ground truth label and the prediction as input and produces a non-negative number (referred to as the *loss*) representing how much error we have made on this particular prediction. Our goal for finding a good parameter vector $\theta^*$ is to minimize the average loss on the set of $N$ training examples.

One assumption that is commonly made in machine learning is that the set of examples $(x_1, y_1), \ldots, (x_N, y_N)$ is *independent and identically distributed*. The word *independent* (Section 6.4.5) means that two data points $(x_i, y_i)$ and $(x_j, y_j)$ do not statistically depend on each other, meaning that the empirical mean is a good estimate of the population mean (Section 6.4.1). This implies that we can use the empirical mean of the loss on the training data.

For a given training set $\{(x_1, y_1), \ldots, (x_N, y_N)\}$, we introduce the notation of an example matrix $X := [x_1, \ldots, x_N]^\top \in \mathbb{R}^{N \times D}$ and a label vector $y := [y_1, \ldots, y_N]^\top \in \mathbb{R}^N$. Using this matrix notation the average loss is given by

$$
R_{\text{emp}}(f, X, y) = \frac{1}{N} \sum_{n=1}^N \ell(y_n, \hat{y}_n), \tag{8.6}
$$

where $\hat{y}_n = f(x_n, \theta)$. Equation (8.6) is called the *empirical risk* and depends on three arguments, the predictor $f$ and the data $X$, $y$. This general strategy for learning is called *empirical risk minimization*.

## Example 2 (Least-Squares Loss)

Continuing the example of least-squares regression, we specify that we measure the cost of making an error during training using the squared loss $\ell(y_n, \hat{y}_n) = (y_n - \hat{y}_n)^2$. We wish to minimize the empirical risk (8.6), which is the average of the losses over the data

$$
\min_{\theta \in \mathbb{R}^D} \frac{1}{N} \sum_{n=1}^N (y_n - f(x_n, \theta))^2, \tag{8.7}
$$

where we substituted the predictor $\hat{y}_n = f(x_n, \theta)$. By using our choice of a linear predictor $f(x_n, \theta) = \theta^\top x_n$, we obtain the optimization problem

$$
\min_{\theta \in \mathbb{R}^D} \frac{1}{N} \sum_{n=1}^N (y_n - \theta^\top x_n)^2. \tag{8.8}
$$

This equation can be equivalently expressed in matrix form

$$
\min_{\theta \in \mathbb{R}^D} \frac{1}{N} \| y - X\theta \|^2. \tag{8.9}
$$

This is known as the *least-squares problem*. There exists a closed-form analytic solution for this by solving the normal equations, which we will discuss in Section 9.2.

We are not interested in a predictor that only performs well on the training data. Instead, we seek a predictor that performs well (has low risk) on unseen test data. More formally, we are interested in finding a predictor $f$ (with parameters fixed) that minimizes the *expected risk*

$$
R_{\text{true}}(f) = \mathbb{E}_{x,y} [\ell(y, f(x))], \tag{8.10}
$$

where $y$ is the label and $f(x)$ is the prediction based on the example $x$. The notation $R_{\text{true}}(f)$ indicates that this is the true risk if we had access to an infinite amount of data. The expectation is over the (infinite) set of all possible data and labels.

> **Note**: Another phrase for the expected risk that is commonly used is the “population risk”.

There are two practical questions that arise from our desire to minimize expected risk, which we address in the following two subsections:
- How should we change our training procedure to generalize well?
- How do we estimate expected risk from (finite) data?

> **Remark**: Many machine learning tasks are specified with an associated performance measure, e.g., accuracy of prediction or root mean squared error. The performance measure could be more complex, be cost sensitive, and capture details about the particular application. In principle, the design of the loss function for empirical risk minimization should correspond directly to the performance measure specified by the machine learning task. In practice, there is often a mismatch between the design of the loss function and the performance measure. This could be due to issues such as ease of implementation or efficiency of optimization. $\diamond$

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
from random import shuffle

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Data from Table 8.2 ---
# Age (x_n) and Annual Salary (y_n in thousands)
data = [
    (36, 89.563),  # Aditya
    (47, 123.543),  # Bob
    (26, 23.989),  # Chloé
    (68, 138.769),  # Daisuke
    (33, 113.888),  # Elisabeth
]

# --- Data Preprocessing ---
def standardize(data_list):
    """Standardize a list to have mean 0 and variance 1."""
    n = len(data_list)
    mean = sum(data_list) / n
    variance = sum((x - mean) ** 2 for x in data_list) / n
    std_dev = math.sqrt(variance) if variance > 0 else 1
    return [(x - mean) / std_dev for x in data_list], mean, std_dev

# Extract x (age) and y (salary)
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]

# Standardize features
x_std, x_mean, x_std_dev = standardize(x_values)
y_std, y_mean, y_std_dev = standardize(y_values)

# Add bias term (x_n^{(0)} = 1) for affine function
x_std_with_bias = [[1, x] for x in x_std]
data_with_bias = list(zip(x_std_with_bias, y_std, x_values, y_values))

# --- Split Data into Training and Test Sets ---
def split_train_test(data, test_ratio=0.2):
    """Split data into training and test sets."""
    shuffle(data)
    n = len(data)
    test_size = int(n * test_ratio)
    test_data = data[:test_size]
    train_data = data[test_size:]
    return train_data, test_data

train_data, test_data = split_train_test(data_with_bias, test_ratio=0.2)
X_train = [d[0] for d in train_data]
y_train = [d[1] for d in train_data]
x_train_orig = [d[2] for d in train_data]
y_train_orig = [d[3] for d in train_data]
X_test = [d[0] for d in test_data]
y_test = [d[1] for d in test_data]
x_test_orig = [d[2] for d in test_data]
y_test_orig = [d[3] for d in test_data]

# --- Least-Squares Regression (Normal Equations) ---
def dot_product(v1, v2):
    """Compute dot product of two vectors."""
    return sum(a * b for a, b in zip(v1, v2))

def matrix_multiply(A, B):
    """Multiply two matrices A (m x n) and B (n x p)."""
    m, n = len(A), len(A[0])
    n_b, p = len(B), len(B[0])
    result = [[0 for _ in range(p)] for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

def transpose_matrix(A):
    """Transpose a matrix."""
    m, n = len(A), len(A[0])
    return [[A[i][j] for i in range(m)] for j in range(n)]

def inverse_2x2(A):
    """Compute the inverse of a 2x2 matrix."""
    det = A[0][0] * A[1][1] - A[0][1] * A[1][0]
    if abs(det) < 1e-10:
        return [[0, 0], [0, 0]]  # Singular matrix approximation
    inv_det = 1 / det
    return [
        [A[1][1] * inv_det, -A[0][1] * inv_det],
        [-A[1][0] * inv_det, A[0][0] * inv_det]
    ]

def fit_least_squares(X, y):
    """
    Fit a linear model by solving the normal equations: (X^T X) theta = X^T y.
    Returns theta (including theta_0).
    """
    # Compute X^T X
    X_T = transpose_matrix(X)
    XTX = matrix_multiply(X_T, X)
    
    # Compute X^T y
    XTy = matrix_multiply(X_T, [[yi] for yi in y])
    XTy = [row[0] for row in XTy]
    
    # Solve (X^T X) theta = X^T y
    # For 2x2 matrix (since D=1 + bias), use direct inverse
    XTX_inv = inverse_2x2(XTX)
    theta = matrix_multiply(XTX_inv, [[XTy[i]] for i in range(len(XTy))])
    theta = [row[0] for row in theta]
    
    return theta

def predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev):
    """Predict y for a given x, unstandardizing the result."""
    x_std = (x - x_mean) / x_std_dev
    x_input = [1, x_std]  # Add bias term
    y_std = dot_product(theta, x_input)
    y = y_std * y_std_dev + y_mean
    return y

# --- Compute Empirical and Test Risk ---
def squared_loss(y_true, y_pred):
    """Compute squared loss."""
    return (y_true - y_pred) ** 2

def compute_risk(X, y, theta):
    """Compute average squared loss (empirical or test risk)."""
    N = len(y)
    predictions = [dot_product(theta, x) for x in X]
    losses = [squared_loss(yi, pred) for yi, pred in zip(y, predictions)]
    return sum(losses) / N

# --- Main Demonstration ---
def demonstrate_least_squares():
    """Demonstrate least-squares regression and risk estimation."""
    print("=== Least-Squares Regression Demonstration ===")
    print("Section 8.2.2: Loss Function for Training\n")

    # Step 1: Fit the model using normal equations
    print("Fitting the Model Using Normal Equations")
    theta = fit_least_squares(X_train, y_train)
    print(f"Fitted parameters: theta = {theta}\n")
    
    # Step 2: Compute empirical risk on training data
    print("Computing Empirical Risk (Equation 8.9)")
    emp_risk = compute_risk(X_train, y_train, theta)
    print(f"Empirical risk on training data: {emp_risk:.4f}\n")
    
    # Step 3: Compute risk on test data (simulating expected risk)
    print("Computing Test Risk (Simulating Expected Risk, Equation 8.10)")
    if test_data:
        test_risk = compute_risk(X_test, y_test, theta)
        print(f"Test risk on unseen data: {test_risk:.4f}\n")
    else:
        print("Test set is empty; cannot compute test risk.\n")
    
    # Step 4: Predict on new data
    print("Prediction on New Data")
    x_query = 60
    y_pred = predict(x_query, theta, x_mean, x_std_dev, y_mean, y_std_dev)
    print(f"Prediction at x = {x_query}: y = {y_pred:.3f} thousand\n")
    
    return theta, x_query, y_pred

# --- Plotting Function ---
def plot_results(theta, x_query, y_pred):
    """Plot the training data, test data, fitted model, and prediction."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_train_orig, y=y_train_orig, color="blue", label="Training data")
    if test_data:
        sns.scatterplot(x=x_test_orig, y=y_test_orig, color="green", label="Test data")
    
    # Plot the linear fit
    x_range = [min(x_values) - 5, max(x_values) + 5]
    y_fit = [predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    sns.lineplot(x=x_range, y=y_fit, color="black", label="Fitted linear model")
    
    # Mark prediction at x = 60
    plt.scatter([x_query], [y_pred], color="red", s=100, marker="*", label=f"Prediction at x = {x_query}")
    plt.title("Least-Squares Regression: Training, Test, and Prediction")
    plt.xlabel("x (age)")
    plt.ylabel("y (salary in thousands)")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Least-Squares Regression Demonstration with Seaborn Plotting")
    print("=" * 60)

    # Demonstrate least-squares regression
    theta, x_query, y_pred = demonstrate_least_squares()

    # Plot the results
    print("Plotting Training Data, Test Data, Fitted Model, and Prediction")
    plot_results(theta, x_query, y_pred)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Fitted model parameters: theta = {theta}")
    print(f"• Predicted salary at age {x_query}: {y_pred:.3f} thousand")
    print("• Plotted training data, test data, fitted model, and prediction")
```

#  Regularization to Reduce Overfitting

This section describes an addition to empirical risk minimization that allows it to generalize well (approximately minimizing expected risk). Recall that the aim of training a machine learning predictor is so that we can perform well on unseen data, i.e., the predictor generalizes well. We simulate this unseen data by holding out a proportion of the whole dataset. This hold out set is referred to as the *test set*.

Given a sufficiently rich class of functions for the predictor $f$, we can essentially memorize the training data to obtain zero empirical risk. While this is great to minimize the loss (and therefore the risk) on the training data, we would not expect the predictor to generalize well to unseen data. In practice, we have only a finite set of data, and hence we split our data into a training and a test set. The training set is used to fit the model, and the test set (not seen by the machine learning algorithm during training) is used to evaluate generalization performance. It is important for the user to not cycle back to a new round of training after having observed the test set. We use the subscripts $\text{train}$ and $\text{test}$ to denote the training and test sets, respectively. We will revisit this idea of using a finite dataset to evaluate expected risk in Section 8.2.4.

It turns out that empirical risk minimization can lead to *overfitting*, i.e., the predictor fits too closely to the training data and does not generalize well to new data (Mitchell, 1997). This general phenomenon of having very small average loss on the training set but large average loss on the test set tends to occur when we have little data and a complex hypothesis class. For a particular predictor $f$ (with parameters fixed), the phenomenon of overfitting occurs when the risk estimate from the training data $R_{\text{emp}}(f, X_{\text{train}}, y_{\text{train}})$ underestimates the expected risk $R_{\text{true}}(f)$. Since we estimate the expected risk $R_{\text{true}}(f)$ by using the empirical risk on the test set $R_{\text{emp}}(f, X_{\text{test}}, y_{\text{test}})$, if the test risk is much larger than the training risk, this is an indication of overfitting. We revisit the idea of overfitting in Section 8.3.3.

Therefore, we need to somehow bias the search for the minimizer of empirical risk by introducing a penalty term, which makes it harder for the optimizer to return an overly flexible predictor. In machine learning, the penalty term is referred to as *regularization*. Regularization is a way to compromise between accurate solution of empirical risk minimization and the size or complexity of the solution.

## Example 8.3 (Regularized Least Squares)

Regularization is an approach that discourages complex or extreme solutions to an optimization problem. The simplest regularization strategy is to replace the least-squares problem

$$
\min_{\theta} \frac{1}{N} \| y - X\theta \|^2, \tag{8.11}
$$

in the previous example with the “regularized” problem by adding a penalty term involving only $\theta$:

$$
\min_{\theta} \frac{1}{N} \| y - X\theta \|^2 + \lambda \| \theta \|^2_2, \tag{8.12}
$$

The additional term $\| \theta \|_2^2$ is called the *regularizer*, and the parameter $\lambda$ is the *regularization parameter*. The regularization parameter trades off minimizing the loss on the training set and the magnitude of the parameters $\theta$. It often happens that the magnitude of the parameter values becomes relatively large if we run into overfitting (Bishop, 2006). The regularization term is sometimes called the *penalty term*, which biases the vector $\theta$ to be closer to the origin. The idea of regularization also appears in probabilistic models as the prior probability of the parameters. Recall from Section 6.6 that for the posterior distribution to be of the same form as the prior distribution, the prior and the likelihood need to be conjugate. We will revisit this idea in Section 8.3.2. We will see in Chapter 12 that the idea of the regularizer is equivalent to the idea of a large margin.

##  Cross-Validation to Assess the Generalization Performance

We mentioned in the previous section that we measure the generalization error by estimating it by applying the predictor on test data. This data is also sometimes referred to as the *validation set*. The validation set is a subset of the available training data that we keep aside. A practical issue with this approach is that the amount of data is limited, and ideally we would use as much of the data available to train the model. This would require us to keep our validation set $V$ small, which then would lead to a noisy estimate (with high variance) of the predictive performance.

One solution to these contradictory objectives (large training set, large validation set) is to use *cross-validation*. $K$-fold cross-validation effectively partitions the data into $K$ chunks, $K-1$ of which form the training set $R$, and the last chunk serves as the validation set $V$ (similar to the idea outlined previously). Cross-validation iterates through (ideally) all combinations of assignments of chunks to $R$ and $V$; see Figure 8.4. This procedure is repeated for all $K$ choices for the validation set, and the performance of the model from the $K$ runs is averaged. We partition our dataset into two sets $D = R \cup V$, such that they do not overlap ($R \cap V = \emptyset$), where $V$ is the validation set, and train our model on $R$. After training, we assess the performance of the predictor $f$ on the ## Figure 8.4 K-fold Cross-Validation

validation set $V$ (e.g., by computing root mean square error (RMSE) of the trained model on the validation set). More precisely, for each partition $k$ the training data $R^{(k)}$ produces a predictor $f^{(k)}$, which is then applied to validation set $V^{(k)}$ to compute the empirical risk $R(f^{(k)}, V^{(k)})$. We cycle through all possible partitionings of validation and training sets and compute the average generalization error of the predictor. Cross-validation approximates the expected generalization error

$$
\mathbb{E}_V [R(f, V)] \approx \frac{1}{K} \sum_{k=1}^K R(f^{(k)}, V^{(k)}), \tag{8.13}
$$

where $R(f^{(k)}, V^{(k)})$ is the risk (e.g., RMSE) on the validation set $V^{(k)}$ for predictor $f^{(k)}$. The approximation has two sources: first, due to the finite training set, which results in not the best possible $f^{(k)}$; and second, due to the finite validation set, which results in an inaccurate estimation of the risk $R(f^{(k)}, V^{(k)})$.

A potential disadvantage of $K$-fold cross-validation is the computational cost of training the model $K$ times, which can be burdensome if the training cost is computationally expensive. In practice, it is often not sufficient to look at the direct parameters alone. For example, we need to explore multiple complexity parameters (e.g., multiple regularization parameters), which may not be direct parameters of the model. Evaluating the quality of the model, depending on these hyperparameters, may result in a number of training runs that is exponential in the number of model parameters. One can use *nested cross-validation* (Section 8.6.1) to search for good hyperparameters.

However, cross-validation is an *embarrassingly parallel* problem, i.e., little effort is needed to separate the problem into a number of parallel tasks. Given sufficient computing resources (e.g., cloud computing, server farms), cross-validation does not require longer than a single performance assessment.

In this section, we saw that empirical risk minimization is based on the following concepts: the hypothesis class of functions, the loss function, and regularization. In Section 8.3, we will see the effect of using a probability distribution to replace the idea of the loss function and regularization.

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
from random import shuffle

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Data from Table 8.2 ---
# Age (x_n) and Annual Salary (y_n in thousands)
data = [
    (36, 89.563),  # Aditya
    (47, 123.543),  # Bob
    (26, 23.989),  # Chloé
    (68, 138.769),  # Daisuke
    (33, 113.888),  # Elisabeth
]

# --- Data Preprocessing ---
def standardize(data_list):
    """Standardize a list to have mean 0 and variance 1."""
    n = len(data_list)
    mean = sum(data_list) / n
    variance = sum((x - mean) ** 2 for x in data_list) / n
    std_dev = math.sqrt(variance) if variance > 0 else 1
    return [(x - mean) / std_dev for x in data_list], mean, std_dev

# Extract x (age) and y (salary)
x_values = [row[0] for row in data]
y_values = [row[1] for row in data]

# Standardize features
x_std, x_mean, x_std_dev = standardize(x_values)
y_std, y_mean, y_std_dev = standardize(y_values)

# Add bias term (x_n^{(0)} = 1) for affine function
x_std_with_bias = [[1, x] for x in x_std]
data_with_bias = list(zip(x_std_with_bias, y_std, x_values, y_values))

# --- Matrix Operations ---
def dot_product(v1, v2):
    """Compute dot product of two vectors."""
    return sum(a * b for a, b in zip(v1, v2))

def matrix_multiply(A, B):
    """Multiply two matrices A (m x n) and B (n x p)."""
    m, n = len(A), len(A[0])
    n_b, p = len(B), len(B[0])
    result = [[0 for _ in range(p)] for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

def transpose_matrix(A):
    """Transpose a matrix."""
    m, n = len(A), len(A[0])
    return [[A[i][j] for i in range(m)] for j in range(n)]

def inverse_2x2(A):
    """Compute the inverse of a 2x2 matrix."""
    det = A[0][0] * A[1][1] - A[0][1] * A[1][0]
    if abs(det) < 1e-10:
        return [[0, 0], [0, 0]]  # Singular matrix approximation
    inv_det = 1 / det
    return [
        [A[1][1] * inv_det, -A[0][1] * inv_det],
        [-A[1][0] * inv_det, A[0][0] * inv_det]
    ]

# --- Regularized Least-Squares Regression ---
def fit_regularized_least_squares(X, y, lambda_reg):
    """
    Fit a linear model with L2 regularization: min (1/N) ||y - X theta||^2 + lambda ||theta||^2.
    Solves (X^T X + N*lambda I) theta = X^T y.
    Returns theta (including theta_0).
    """
    N = len(X)
    # Compute X^T X
    X_T = transpose_matrix(X)
    XTX = matrix_multiply(X_T, X)
    
    # Add regularization term: X^T X + N*lambda I (regularize only non-bias terms)
    for i in range(1, len(XTX)):
        XTX[i][i] += N * lambda_reg
    
    # Compute X^T y
    XTy = matrix_multiply(X_T, [[yi] for yi in y])
    XTy = [row[0] for row in XTy]
    
    # Solve (X^T X + N*lambda I) theta = X^T y
    XTX_inv = inverse_2x2(XTX)
    theta = matrix_multiply(XTX_inv, [[XTy[i]] for i in range(len(XTy))])
    theta = [row[0] for row in theta]
    
    return theta

def predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev):
    """Predict y for a given x, unstandardizing the result."""
    x_std = (x - x_mean) / x_std_dev
    x_input = [1, x_std]  # Add bias term
    y_std = dot_product(theta, x_input)
    y = y_std * y_std_dev + y_mean
    return y

# --- K-Fold Cross-Validation ---
def squared_loss(y_true, y_pred):
    """Compute squared loss."""
    return (y_true - y_pred) ** 2

def compute_risk(X, y, theta):
    """Compute average squared loss (empirical risk)."""
    N = len(y)
    predictions = [dot_product(theta, x) for x in X]
    losses = [squared_loss(yi, pred) for yi, pred in zip(y, predictions)]
    return sum(losses) / N

def k_fold_cross_validation(data, K, lambda_reg):
    """
    Perform K-fold cross-validation to estimate generalization error (Equation 8.13).
    Returns average validation risk.
    """
    shuffle(data)
    fold_size = len(data) // K
    risks = []
    
    for k in range(K):
        # Split into training (R) and validation (V) sets
        val_start = k * fold_size
        val_end = (k + 1) * fold_size if k < K - 1 else len(data)
        V = data[val_start:val_end]  # Validation set
        R = data[:val_start] + data[val_end:]  # Training set
        
        # Extract X and y
        X_train = [d[0] for d in R]
        y_train = [d[1] for d in R]
        X_val = [d[0] for d in V]
        y_val = [d[1] for d in V]
        
        # Fit model on training data
        theta = fit_regularized_least_squares(X_train, y_train, lambda_reg)
        
        # Compute risk on validation set
        risk = compute_risk(X_val, y_val, theta)
        risks.append(risk)
    
    # Average risk over K folds (Equation 8.13)
    return sum(risks) / K

# --- Hyperparameter Tuning with Cross-Validation ---
def select_lambda(data, K, lambda_candidates):
    """Select the best regularization parameter using K-fold cross-validation."""
    risks = []
    for lambda_reg in lambda_candidates:
        avg_risk = k_fold_cross_validation(data, K, lambda_reg)
        risks.append((avg_risk, lambda_reg))
    best_lambda = min(risks, key=lambda x: x[0])[1]
    return best_lambda

# --- Main Demonstration ---
def demonstrate_regularization_and_cv():
    """Demonstrate regularization and cross-validation for least-squares regression."""
    print("=== Regularization and Cross-Validation Demonstration ===")
    print("Sections 8.2.3 and 8.2.4\n")

    # Step 1: Select regularization parameter using 5-fold cross-validation
    print("Selecting Regularization Parameter with 5-Fold Cross-Validation")
    K = 5
    lambda_candidates = [0, 0.01, 0.1, 0.5, 1.0]
    best_lambda = select_lambda(data_with_bias, K, lambda_candidates)
    print(f"Best regularization parameter (lambda): {best_lambda}\n")
    
    # Step 2: Fit the model on all data with best lambda
    print("Fitting the Model with Regularization (Equation 8.12)")
    X = [d[0] for d in data_with_bias]
    y = [d[1] for d in data_with_bias]
    theta = fit_regularized_least_squares(X, y, best_lambda)
    print(f"Fitted parameters: theta = {theta}\n")
    
    # Step 3: Compute empirical risk on all data
    print("Computing Empirical Risk")
    emp_risk = compute_risk(X, y, theta)
    print(f"Empirical risk on all data: {emp_risk:.4f}\n")
    
    # Step 4: Predict on new data
    print("Prediction on New Data")
    x_query = 60
    y_pred = predict(x_query, theta, x_mean, x_std_dev, y_mean, y_std_dev)
    print(f"Prediction at x = {x_query}: y = {y_pred:.3f} thousand\n")
    
    return theta, x_query, y_pred

# --- Plotting Function ---
def plot_results(theta, x_query, y_pred):
    """Plot the training data, fitted model, and prediction."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_values, y=y_values, color="blue", label="Training data")
    
    # Plot the linear fit
    x_range = [min(x_values) - 5, max(x_values) + 5]
    y_fit = [predict(x, theta, x_mean, x_std_dev, y_mean, y_std_dev) for x in x_range]
    sns.lineplot(x=x_range, y=y_fit, color="black", label="Fitted linear model")
    
    # Mark prediction at x = 60
    plt.scatter([x_query], [y_pred], color="red", s=100, marker="*", label=f"Prediction at x = {x_query}")
    plt.title("Regularized Least-Squares Regression with Cross-Validation")
    plt.xlabel("x (age)")
    plt.ylabel("y (salary in thousands)")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Regularization and Cross-Validation Demonstration with Seaborn Plotting")
    print("=" * 60)

    # Demonstrate regularization and cross-validation
    theta, x_query, y_pred = demonstrate_regularization_and_cv()

    # Plot the results
    print("Plotting Training Data, Fitted Model, and Prediction")
    plot_results(theta, x_query, y_pred)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Fitted model parameters: theta = {theta}")
    print(f"• Predicted salary at age {x_query}: {y_pred:.3f} thousand")
    print("• Plotted training data, fitted model, and prediction")
```
