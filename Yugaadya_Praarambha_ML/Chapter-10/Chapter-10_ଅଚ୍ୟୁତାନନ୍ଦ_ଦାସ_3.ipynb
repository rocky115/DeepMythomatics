## Latent-Variable Perspective of GMMs

We can interpret the **Gaussian Mixture Model (GMM)** using a **discrete latent-variable model** perspective, where the latent variable $ z $ takes on one of $ K $ finite values. This contrasts with **PCA**, where the latent variables are continuous in $ \mathbb{R}^M $.

## Advantages of the Probabilistic Perspective
1. Justifies previous heuristic decisions.
2. Interprets responsibilities $ r_{nk} $ as posterior probabilities.
3. Derives the **EM algorithm** from principled maximum likelihood estimation in latent-variable models.

---

## Generative Process and Probabilistic Model

To define a probabilistic model for GMMs, consider the following generative process:

- There are $ K $ mixture components.
- A data point $ x $ is generated by exactly one component.
- Let the indicator variable $ z_k \in \{0, 1\} $ represent whether component $ k $ generated the point:

$$
p(x \mid z_k = 1) = \mathcal{N}(x \mid \mu_k, \Sigma_k) \tag{11.58}
$$

Define the **latent vector**:

$$
z = [z_1, \dots, z_K]^T \in \mathbb{R}^K
$$

with exactly one 1 and the rest 0s (a **1-of-K encoding**). For example, if $ K = 3 $, then a valid $ z $ could be:

$$
z = [0, 1, 0]^T
$$

### Remark:
This form of distribution is called a **Multinoulli distribution**, a generalization of the Bernoulli distribution to multiple values.

---

## Prior Over Latent Variables

We place a categorical prior $ p(z) $ over the latent variables:

$$
p(z) = \pi = [\pi_1, \dots, \pi_K]^T, \quad \sum_{k=1}^K \pi_k = 1 \tag{11.59}
$$

Thus:

$$
\pi_k = p(z_k = 1) \tag{11.60}
$$

---

## Sampling from a GMM (Ancestral Sampling)

1. Sample $ z^{(i)} \sim p(z) $
2. Sample $ x^{(i)} \sim p(x \mid z^{(i)} = 1) $

By ignoring the sampled $ z^{(i)} $, we obtain a sample from the marginal distribution $ p(x) $.

---

## Joint Distribution

The joint distribution over $ x $ and $ z $ is defined as:

$$
p(x, z_k = 1) = p(x \mid z_k = 1) \cdot p(z_k = 1) = \pi_k \cdot \mathcal{N}(x \mid \mu_k, \Sigma_k) \tag{11.61}
$$

Thus, for all $ k = 1, \dots, K $:

$$
p(x, z) =
\begin{bmatrix}
\pi_1 \mathcal{N}(x \mid \mu_1, \Sigma_1) \\
\vdots \\
\pi_K \mathcal{N}(x \mid \mu_K, \Sigma_K)
\end{bmatrix}
\tag{11.62}
$$

This fully specifies the GMM as a **probabilistic latent-variable model**.


```python
import random
import math
import matplotlib.pyplot as plt

# Define GMM parameters
K = 3
pi = [0.3, 0.4, 0.3]
mus = [-2.5, 0.0, 3.0]
sigmas = [0.5, 1.0, 0.8]

def sample_discrete(pi):
    r = random.random()
    total = 0
    for k, p in enumerate(pi):
        total += p
        if r <= total:
            return k
    return len(pi) - 1

def gaussian_sample(mu, sigma):
    u1, u2 = random.random(), random.random()
    z = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    return mu + sigma * z

def gaussian_pdf(x, mu, sigma):
    coeff = 1.0 / (math.sqrt(2 * math.pi) * sigma)
    exp_val = -((x - mu) ** 2) / (2 * sigma ** 2)
    return coeff * math.exp(exp_val)

def sample_gmm(n_samples):
    samples = []
    labels = []
    for _ in range(n_samples):
        k = sample_discrete(pi)
        x = gaussian_sample(mus[k], sigmas[k])
        samples.append(x)
        labels.append(k)
    return samples, labels

# Generate data
n_samples = 1000
samples, labels = sample_gmm(n_samples)

# Plot histogram and densities
x_vals = [x / 100.0 for x in range(-500, 800)]  # x in [-5, 8]
mixture_density = [sum(pi[k] * gaussian_pdf(x, mus[k], sigmas[k]) for k in range(K)) for x in x_vals]

plt.figure(figsize=(10, 5))

# Histogram of GMM samples
plt.hist(samples, bins=40, density=True, alpha=0.6, color='gray', label='Sample histogram')

# Plot GMM density
plt.plot(x_vals, mixture_density, label='GMM density', color='black')

# Plot individual Gaussian components
for k in range(K):
    component_density = [pi[k] * gaussian_pdf(x, mus[k], sigmas[k]) for x in x_vals]
    plt.plot(x_vals, component_density, linestyle='--', label=f'Component {k+1}')

plt.title("Sampled GMM and Component Densities")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Likelihood

To obtain the likelihood $p(x \mid \theta)$ in a latent-variable model, we marginalize out the latent variables:

$$
p(x \mid \theta) = \sum_{z} p(x \mid \theta, z) p(z \mid \theta), \quad \theta := \{\mu_k, \Sigma_k, \pi_k : k = 1, \dots, K\} \tag{11.63}
$$

Since $z$ is a one-hot vector, we sum over $K$ possible values:

$$
p(x \mid \theta) = \sum_{k=1}^K p(x \mid \theta, z_k = 1)p(z_k = 1 \mid \theta) \tag{11.65}
$$

Resulting in:

$$
p(x \mid \theta) = \sum_{k=1}^K \pi_k \, \mathcal{N}(x \mid \mu_k, \Sigma_k) \tag{11.66}
$$

Given a dataset $X = \{x_1, \dots, x_N\}$, the full likelihood becomes:

$$
p(X \mid \theta) = \prod_{n=1}^N p(x_n \mid \theta) = \prod_{n=1}^N \sum_{k=1}^K \pi_k \, \mathcal{N}(x_n \mid \mu_k, \Sigma_k) \tag{11.67}
$$

This is the likelihood used in Gaussian Mixture Models (GMMs).

---

## Posterior Distribution

Using Bayes' rule, the posterior of component $k$ given data point $x$ is:

$$
p(z_k = 1 \mid x) = \frac{p(z_k = 1)p(x \mid z_k = 1)}{p(x)} \tag{11.68}
$$

Substituting from earlier equations:

$$
p(z_k = 1 \mid x) = \frac{\pi_k \, \mathcal{N}(x \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \, \mathcal{N}(x \mid \mu_j, \Sigma_j)} \tag{11.69}
$$

This is the **responsibility** $r_{nk}$ of component $k$ for data point $x$.

---

## Extension to a Full Dataset

For a dataset $X = \{x_1, \dots, x_N\}$, each data point $x_n$ has its own latent variable:

$$
z^{(n)} = [z_{n1}, \dots, z_{nK}]^\top \in \mathbb{R}^K \tag{11.70}
$$

The latent variables share the same prior $\pi$. The conditional distribution factorizes:

$$
p(x_1, \dots, x_N \mid z^{(1)}, \dots, z^{(N)}) = \prod_{n=1}^N p(x_n \mid z^{(n)}) \tag{11.71}
$$

Applying Bayes' theorem again for each data point:

$$
p(z_{nk} = 1 \mid x_n) = \frac{p(x_n \mid z_{nk} = 1)p(z_{nk} = 1)}{\sum_{j=1}^K p(x_n \mid z_{nj} = 1)p(z_{nj} = 1)} \tag{11.72a}
$$

Simplified as:

$$
r_{nk} = \frac{\pi_k \, \mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \, \mathcal{N}(x_n \mid \mu_j, \Sigma_j)} \tag{11.72b}
$$

So the responsibilities have a clear probabilistic interpretation as posterior probabilities.


```python
import math
import matplotlib.pyplot as plt

# Gaussian density function
def gaussian(x, mu, sigma2):
    coeff = 1.0 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu)**2) / (2 * sigma2)
    return coeff * math.exp(exponent)

# Likelihood p(x|θ)
def p_x_given_theta(x, pis, mus, sigmas2):
    return sum(pis[k] * gaussian(x, mus[k], sigmas2[k]) for k in range(len(pis)))

# Responsibilities γ(z_k = 1 | x)
def responsibilities(X, pis, mus, sigmas2):
    R = []
    for x in X:
        total = sum(pis[k] * gaussian(x, mus[k], sigmas2[k]) for k in range(len(pis)))
        r_x = [(pis[k] * gaussian(x, mus[k], sigmas2[k])) / (total + 1e-12) for k in range(len(pis))]
        R.append(r_x)
    return R

# Example data and parameters
X = [-2.5, -1, 0, 2, 4]
pis = [0.3, 0.4, 0.3]
mus = [-2.0, 0.0, 3.5]
sigmas2 = [0.25, 1.0, 0.64]  # σ² = σ²

# Plot GMM density
x_vals = [x / 100.0 for x in range(-600, 800)]
densities = [p_x_given_theta(x, pis, mus, sigmas2) for x in x_vals]

plt.figure(figsize=(10, 4))

# Plot total GMM likelihood curve
plt.subplot(1, 2, 1)
plt.plot(x_vals, densities, label="GMM density", color="black")
plt.scatter(X, [p_x_given_theta(x, pis, mus, sigmas2) for x in X], color="red", label="Data points")
plt.title("GMM Likelihood Curve")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

# Plot responsibilities for each component
R = responsibilities(X, pis, mus, sigmas2)
plt.subplot(1, 2, 2)
x_idx = list(range(len(X)))
for k in range(len(pis)):
    plt.plot(x_idx, [r[k] for r in R], marker='o', label=f'Component {k+1}')
plt.title("Responsibilities γ(z_k | x)")
plt.xlabel("Data Point Index")
plt.ylabel("Responsibility")
plt.xticks(x_idx, [str(x) for x in X])
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```



## EM Algorithm Revisited

The Expectation-Maximization (EM) algorithm can be derived from the **latent-variable perspective**. Given a current parameter estimate $\theta^{(t)}$, the E-step computes the expected complete log-likelihood:

$$
Q(\theta \mid \theta^{(t)}) = \mathbb{E}_{z \mid x, \theta^{(t)}} \left[\log p(x, z \mid \theta)\right] \tag{11.73a}
$$

Equivalently:

$$
Q(\theta \mid \theta^{(t)}) = \int \log p(x, z \mid \theta) \, p(z \mid x, \theta^{(t)}) \, dz \tag{11.73b}
$$

The M-step finds updated parameters:

$$
\theta^{(t+1)} = \arg\max_\theta Q(\theta \mid \theta^{(t)})
$$

Although each EM iteration increases the log-likelihood, there is no guarantee of reaching a **global maximum**—only a **local optimum**. Different initializations of $\theta$ can lead to different solutions.

---

## Sampling from a GMM

To generate data from a GMM:
1. Sample $k$ from $\pi = [\pi_1, \ldots, \pi_K]^\top$
2. Sample $x \sim \mathcal{N}(\mu_k, \Sigma_k)$

Repeating this $N$ times yields synthetic GMM data.

---

## Relationship to K-means

- GMMs use **soft assignments** (responsibilities $r_{nk}$).
- K-means uses **hard assignments** by minimizing squared Euclidean distances.
- If all $\Sigma_k = I$, and $\pi_k$ are uniform, GMM reduces to soft K-means.

---

## Estimation and Model Selection

Maximum likelihood estimation can lead to **overfitting**:
- When a component mean matches a data point exactly,
- And the covariance $\Sigma_k \to 0$, 
- Then $p(x \mid \theta) \to \infty$.

To avoid this:
- Use **Bayesian inference** by placing priors on $\pi_k, \mu_k, \Sigma_k$
- Compute **posterior distribution** over parameters
- Estimate **model evidence** for model comparison
- If closed-form is intractable, use **variational inference**

## Nonparametric Density Estimation

## Histograms

Histograms represent density by **counting** data in bins:

- Proposed by Pearson (1895)
- Sensitive to bin width $h$
- Overfitting/underfitting risks

---

## Kernel Density Estimation (KDE)

Proposed by Rosenblatt (1956) and Parzen (1962):

$$
p(x) = \frac{1}{N h} \sum_{n=1}^N k\left(\frac{x - x_n}{h}\right) \tag{11.74}
$$

- $k$ is a **kernel function**: non-negative and integrates to 1
- $h$ is the **bandwidth**: controls smoothness

### Common Kernels:
- Uniform
- Gaussian

KDE ensures smooth density estimates and is closely related to histograms.

---

## Illustration (Fig.13)

- Histogram (orange bars): Unsmoothed counts
- KDE (blue curve): Smoothed density using Gaussian kernels

The **KDE provides a smoother estimate** of the underlying distribution than a histogram.




```python
import math
import random
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
random.seed(42)
data = [random.gauss(2, 1.0) for _ in range(150)] + [random.gauss(6, 0.7) for _ in range(100)]

# Step 2: Histogram (manual binning)
def histogram(data, bin_width):
    min_x = min(data)
    max_x = max(data)
    bins = {}
    for x in data:
        bin_idx = int((x - min_x) / bin_width)
        bins[bin_idx] = bins.get(bin_idx, 0) + 1
    bin_edges = [(min_x + i * bin_width) for i in sorted(bins)]
    heights = [bins[i] / (len(data) * bin_width) for i in sorted(bins)]
    return bin_edges, heights

# Step 3: KDE using Gaussian kernel
def gaussian_kernel(x, xi, h):
    return (1 / (math.sqrt(2 * math.pi) * h)) * math.exp(-((x - xi) ** 2) / (2 * h ** 2))

def kde(data, x_vals, h):
    kde_vals = []
    N = len(data)
    for x in x_vals:
        val = sum(gaussian_kernel(x, xi, h) for xi in data) / N
        kde_vals.append(val)
    return kde_vals

# Parameters
bin_width = 0.5
bandwidth = 0.4
x_vals = [x / 100.0 for x in range(int(100 * (min(data) - 1)), int(100 * (max(data) + 1)))]

# Compute histogram and KDE
hist_x, hist_y = histogram(data, bin_width)
kde_y = kde(data, x_vals, bandwidth)

# Plot
plt.figure(figsize=(10, 5))

# Histogram
plt.bar(hist_x, hist_y, width=bin_width, align='edge', alpha=0.6, color='orange', label='Histogram')

# KDE
plt.plot(x_vals, kde_y, color='blue', linewidth=2, label='KDE (Gaussian)')

plt.title("Histogram vs Kernel Density Estimation")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
