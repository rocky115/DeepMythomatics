##  Traditional Derivation of the Margin

In Section 12.2.1, we assumed $\|w\| = 1$ to focus on direction only. Here, we take a different approach: we scale the data so that the closest example $x_a$ satisfies:

$$
\langle w, x_a \rangle + b = 1
$$

Since $x_a'$ is the orthogonal projection of $x_a$ onto the hyperplane, by definition:

$$
\langle w, x_a' \rangle + b = 0 \tag{12.11}
$$

Substituting the vector decomposition $x_a = x_a' + \frac{r}{\|w\|}w$ into the equation:

$$
\left\langle w, x_a - \frac{r}{\|w\|}w \right\rangle + b = 0 \tag{12.12}
$$

By bilinearity of the inner product:

$$
\langle w, x_a \rangle + b - r \|w\| = 0 \tag{12.13}
$$

Since $\langle w, x_a \rangle + b = 1$ by assumption, we solve for $r$:

$$
r = \frac{1}{\|w\|} \tag{12.14}
$$

---

## Margin Constraints

We require that **all** examples be at least distance $1$ away from the hyperplane on the correct side. This yields:

$$
y_n(\langle w, x_n \rangle + b) \geq 1 \tag{12.15}
$$

---

## Optimization Formulation

Maximizing the margin is equivalent to minimizing the norm of $w$:

$$
\begin{aligned}
\min_{w,b} \quad & \frac{1}{\|w\|} \tag{12.16} \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1 \quad \forall n \tag{12.17}
\end{aligned}
$$

Instead of maximizing the reciprocal, we minimize the squared norm (which is convex):

$$
\begin{aligned}
\min_{w,b} \quad & \frac{1}{2} \|w\|^2 \tag{12.18} \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1 \quad \forall n \tag{12.19}
\end{aligned}
$$

This is the standard formulation of the **hard margin SVM**, which does not allow for margin violations.


```python
import random
import matplotlib.pyplot as plt

# Generate linearly separable dataset
def generate_data(n=50):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        label = 1 if x1 + x2 > 0.5 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product
def dot(w, x):
    return sum(wi * xi for wi, xi in zip(w, x))

# Gradient descent for hard-margin SVM
def train_svm(X, y, lr=0.01, epochs=1000, lambda_reg=0.01):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < 1:
                w = [wi - lr * (2 * lambda_reg * wi - yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * yi
            else:
                w = [wi - lr * (2 * lambda_reg * wi) for wi in w]
    return w, b

# Visualize decision boundary and margin
def plot_svm(X, y, w, b):
    import math
    def norm(v): return math.sqrt(dot(v, v))

    plt.figure(figsize=(6, 6))
    for xi, yi in zip(X, y):
        plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')

    x_vals = [-2.5, 2.5]
    y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
    margin = 1 / norm(w)
    y_margin_up = [y + margin for y in y_vals]
    y_margin_down = [y - margin for y in y_vals]

    plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
    plt.plot(x_vals, y_margin_up, 'r--', alpha=0.6, label='Margin')
    plt.plot(x_vals, y_margin_down, 'r--', alpha=0.6)
    plt.title("Hard-Margin SVM: Traditional Margin Scaling")
    plt.xlabel("x₁")
    plt.ylabel("x₂")
    plt.axis("equal")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run it all
X, y = generate_data()
w, b = train_svm(X, y)
plot_svm(X, y, w, b)
```

# 12.2.3 Why We Can Set the Margin to 1

In Section 12.2.1, we proposed maximizing a margin value $r$, representing the distance of the closest example to the hyperplane under the constraint $\|w\| = 1$.

In Section 12.2.2, we instead scaled the data so that the margin becomes $1$ for the closest point.

This section proves the two formulations are equivalent.

---

## Theorem 12.1

Maximizing the margin $r$ with normalized weights:

$$
\begin{aligned}
\max_{w, b, r} \quad & r \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq r, \quad \|w\| = 1, \quad r > 0
\end{aligned} \tag{12.20}
$$

is equivalent to minimizing the squared norm of $w$ with unit margin constraint:

$$
\begin{aligned}
\min_{w, b} \quad & \frac{1}{2} \|w\|^2 \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1
\end{aligned} \tag{12.21}
$$

---


Fig.6 (a) Linearly separable and (b) non-linearly separable data.

## Proof Sketch

Start with the original problem (12.20). Since squaring is monotonic for non-negative inputs, we can equivalently maximize $r^2$:

$$
\max_{w, b, r} \quad r^2 \quad \text{subject to} \quad y_n(\langle w, x_n \rangle + b) \geq r, \; \|w\| = 1, \; r > 0
$$

Now, reparameterize the weight vector $w'$ without the normalization:

$$
\max_{w', b, r} \quad r^2 \quad \text{subject to} \quad y_n \left( \frac{\langle w', x_n \rangle + b}{\|w'\|} \right) \geq r, \; r > 0 \tag{12.22}
$$

Divide both sides of the inequality by $r > 0$:

$$
y_n \left( \frac{\langle w', x_n \rangle + b}{r \|w'\|} \right) \geq 1 \tag{12.23}
$$

Now define:

$$
w'' = \frac{w'}{r \|w'\|}, \quad b'' = \frac{b}{r \|w'\|} \tag{12.24}
$$

Substituting into the inequality:

$$
y_n(\langle w'', x_n \rangle + b'') \geq 1
$$

which matches the constraint in (12.21). Since

$$
\|w''\| = \frac{1}{r}
$$

maximizing $r$ is equivalent to minimizing $\|w''\|$, or more commonly:

$$
\min_{w'', b''} \quad \frac{1}{2} \|w''\|^2 \tag{12.25}
$$

which concludes the proof.


```python
import random
import matplotlib.pyplot as plt
import math

# Generate separable 2D data
def generate_data(n=50):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        label = 1 if x1 + x2 > 0.5 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product
def dot(w, x): return sum(wi * xi for wi, xi in zip(w, x))

# Norm of vector
def norm(w): return math.sqrt(dot(w, w))

# Unit norm version: maximize r s.t. ||w|| = 1
def train_fixed_norm(X, y, lr=0.1, epochs=200):
    w = [1.0, 1.0]
    b = 0.0
    r = 0.1
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < r:
                b += lr * yi
                r -= lr * 0.01
        w = [wi / norm(w) for wi in w]  # normalize w
    return w, b

# Standard SVM: minimize 1/2||w||² s.t. margin = 1
def train_unit_margin(X, y, lr=0.01, epochs=1000, lam=0.01):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < 1:
                w = [wi - lr * (2 * lam * wi - yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * yi
            else:
                w = [wi - lr * 2 * lam * wi for wi in w]
    return w, b

# Plotting helper
def plot_models(X, y, models, titles):
    plt.figure(figsize=(10, 5))
    for idx, (w, b) in enumerate(models):
        plt.subplot(1, 2, idx+1)
        for xi, yi in zip(X, y):
            plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')
        x_vals = [-2.5, 2.5]
        y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
        plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
        plt.title(titles[idx])
        plt.axis("equal")
        plt.grid(True)
        plt.legend()
    plt.tight_layout()
    plt.show()

# Run both methods
X, y = generate_data()
w1, b1 = train_fixed_norm(X, y)
w2, b2 = train_unit_margin(X, y)

plot_models(X, y, [(w1, b1), (w2, b2)], ["Formulation: max margin (unit norm)", "Formulation: min norm (unit margin)"])
```

## Soft Margin SVM: Geometric View

In cases where the data is **not linearly separable**, we allow some examples to lie **within the margin** or even on the **wrong side** of the hyperplane. This approach leads to the **soft margin SVM**.

---

## Geometric Intuition

For each example-label pair $(x_n, y_n)$, we introduce a **slack variable** $\xi_n \geq 0$, which allows a point to:

- Fall within the margin, or
- Be misclassified (on the wrong side of the hyperplane).

We define the decision boundary as:

$$
f(x) = \langle w, x \rangle + b
$$

and adjust the margin constraint from:

$$
y_n (\langle w, x_n \rangle + b) \geq 1
$$

to:

$$
y_n (\langle w, x_n \rangle + b) \geq 1 - \xi_n \tag{12.26b}
$$

To keep $\xi_n$ meaningful, we constrain:

$$
\xi_n \geq 0 \tag{12.26c}
$$

---

## Soft Margin SVM Optimization Problem

We now minimize a combination of:

- **Margin penalty**: $\|w\|^2$
- **Total slack**: $\sum \xi_n$

The resulting **soft margin SVM** optimization problem is:

$$
\min_{w, b, \{\xi_n\}} \quad \frac{1}{2} \|w\|^2 + C \sum_{n=1}^N \xi_n \tag{12.26a}
$$

subject to:

$$
y_n (\langle w, x_n \rangle + b) \geq 1 - \xi_n \quad \text{for } n = 1, \dots, N \tag{12.26b}
$$

$$
\xi_n \geq 0 \quad \text{for } n = 1, \dots, N \tag{12.26c}
$$

---

### Remarks

- The hyperparameter $C > 0$ controls the **tradeoff** between margin size and classification accuracy on training data.
- A **large $C$** places more emphasis on minimizing classification errors (less regularization).
- The term $\|w\|^2$ is often called the **regularizer**, and $C$ is the **regularization parameter**.
- Note: In this formulation, $b$ is *not* regularized, which simplifies optimization but complicates theoretical analysis.

This formulation is sometimes called the **C-SVM**, especially in libraries like LIBSVM and sklearn.

---

In Section 12.2.5, we’ll reformulate the soft margin SVM from the **loss function perspective**, connecting it to empirical risk minimization.


```python
import random
import matplotlib.pyplot as plt

# Generate synthetic 2D data (some overlap)
def generate_data(n=100):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        noise = random.gauss(0, 0.5)
        label = 1 if x1 + x2 + noise > 0 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product and norm helpers
def dot(w, x): return sum(wi * xi for wi, xi in zip(w, x))
def norm(w): return sum(wi ** 2 for wi in w) ** 0.5

# Train Soft Margin SVM with gradient descent (hinge loss + L2 regularization)
def train_soft_svm(X, y, lr=0.01, epochs=1000, C=1.0):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            margin = yi * (dot(w, xi) + b)
            if margin < 1:
                w = [wi - lr * (2 * wi - C * yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * C * yi
            else:
                w = [wi - lr * 2 * wi for wi in w]
    return w, b

# Plot data, hyperplane, and margins
def plot_svm(X, y, w, b):
    plt.figure(figsize=(6, 6))
    for xi, yi in zip(X, y):
        plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')

    # Decision boundary
    x_vals = [-2.5, 2.5]
    y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
    margin = 1 / (norm(w) + 1e-12)
    y_margin_up = [y + margin for y in y_vals]
    y_margin_down = [y - margin for y in y_vals]

    plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
    plt.plot(x_vals, y_margin_up, 'r--', alpha=0.6)
    plt.plot(x_vals, y_margin_down, 'r--', alpha=0.6)
    plt.title("Soft Margin SVM (Geometric View)")
    plt.xlabel("x₁")
    plt.ylabel("x₂")
    plt.axis("equal")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the pipeline
X, y = generate_data()
w, b = train_soft_svm(X, y, C=10.0)
plot_svm(X, y, w, b)
```
