## Updating the Covariances

### Theorem 2: Updates of the GMM Covariances

The update rule for the covariance parameters $ \Sigma_k $ ($ k = 1, \ldots, K $) of the GMM is given by:

$$
\Sigma_k^{\text{new}} = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top, \tag{11.30}
$$

where $ r_{nk} $ is the responsibility of the $ k $-th component for the $ n $-th data point (Equation 11.17), and $ N_k $ is the total responsibility of the $ k $-th component (Equation 11.24):

$$
N_k = \sum_{n=1}^N r_{nk}.
$$

### Proof of Theorem 2

To derive the update rule, we compute the partial derivative of the log-likelihood $ L $ with respect to the covariance $ \Sigma_k $, set it to zero, and solve for $ \Sigma_k $. We start with the gradient of the log-likelihood:

$$
\frac{\partial L}{\partial \Sigma_k} = \sum_{n=1}^N \frac{\partial \log p(x_n \mid \theta)}{\partial \Sigma_k} = \sum_{n=1}^N \frac{1}{p(x_n \mid \theta)} \frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k}, \tag{11.31}
$$

where $ \frac{1}{p(x_n \mid \theta)} $ is given by Equation (11.16):

$$
\frac{1}{p(x_n \mid \theta)} = \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)}.
$$

Next, we compute $ \frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} $. From the definition of the GMM (Equation 11.9), we isolate the $ k $-th term since only the $ k $-th component depends on $ \Sigma_k $:

$$
\frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} = \frac{\partial}{\partial \Sigma_k} \left[ \pi_k (2\pi)^{-\frac{D}{2}} \det(\Sigma_k)^{-\frac{1}{2}} \exp\left(-\frac{1}{2} (x_n - \mu_k)^\top \Sigma_k^{-1} (x_n - \mu_k)\right) \right], \tag{11.32a}
$$

which simplifies to:

$$
\frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} = \pi_k \frac{\partial}{\partial \Sigma_k} \left[ (2\pi)^{-\frac{D}{2}} \det(\Sigma_k)^{-\frac{1}{2}} \exp\left(-\frac{1}{2} (x_n - \mu_k)^\top \Sigma_k^{-1} (x_n - \mu_k)\right) \right]. \tag{11.32b}
$$

Applying the product rule, we get:

$$
\frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} = \pi_k (2\pi)^{-\frac{D}{2}} \left[ \frac{\partial}{\partial \Sigma_k} \left( \det(\Sigma_k)^{-\frac{1}{2}} \right) \exp\left(-\frac{1}{2} (x_n - \mu_k)^\top \Sigma_k^{-1} (x_n - \mu_k)\right) + \det(\Sigma_k)^{-\frac{1}{2}} \frac{\partial}{\partial \Sigma_k} \exp\left(-\frac{1}{2} (x_n - \mu_k)^\top \Sigma_k^{-1} (x_n - \mu_k)\right) \right]. \tag{11.32c}
$$

We use the following identities (from Sections 5.101 and 5.103):

$$
\frac{\partial}{\partial \Sigma_k} \det(\Sigma_k)^{-\frac{1}{2}} = -\frac{1}{2} \det(\Sigma_k)^{-\frac{1}{2}} \Sigma_k^{-1}, \tag{11.33}
$$

$$
\frac{\partial}{\partial \Sigma_k} \left[ (x_n - \mu_k)^\top \Sigma_k^{-1} (x_n - \mu_k) \right] = -\Sigma_k^{-1} (x_n - \mu_k)(x_n - \mu_k)^\top \Sigma_k^{-1}. \tag{11.34}
$$

Substituting these into Equation (11.32c) and simplifying, we obtain:

$$
\frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} = \pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k) \left[ -\frac{1}{2} \left( \Sigma_k^{-1} - \Sigma_k^{-1} (x_n - \mu_k)(x_n - \mu_k)^\top \Sigma_k^{-1} \right) \right]. \tag{11.35}
$$

Now, substitute Equation (11.35) into Equation (11.31):

$$
\frac{\partial L}{\partial \Sigma_k} = \sum_{n=1}^N \frac{1}{p(x_n \mid \theta)} \frac{\partial p(x_n \mid \theta)}{\partial \Sigma_k} = \sum_{n=1}^N \frac{\pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)} \left[ -\frac{1}{2} \left( \Sigma_k^{-1} - \Sigma_k^{-1} (x_n - \mu_k)(x_n - \mu_k)^\top \Sigma_k^{-1} \right) \right], \tag{11.36a}
$$

where the fraction is the responsibility $ r_{nk} $ (Equation 11.17):

$$
\frac{\partial L}{\partial \Sigma_k} = \sum_{n=1}^N r_{nk} \left[ -\frac{1}{2} \left( \Sigma_k^{-1} - \Sigma_k^{-1} (x_n - \mu_k)(x_n - \mu_k)^\top \Sigma_k^{-1} \right) \right], \tag{11.36b}
$$

which simplifies to:

$$
\frac{\partial L}{\partial \Sigma_k} = -\frac{1}{2} \sum_{n=1}^N r_{nk} \left( \Sigma_k^{-1} - \Sigma_k^{-1} (x_n - \mu_k)(x_n - \mu_k)^\top \Sigma_k^{-1} \right), \tag{11.36c}
$$

or:

$$
\frac{\partial L}{\partial \Sigma_k} = -\frac{1}{2} \Sigma_k^{-1} \sum_{n=1}^N r_{nk} + \frac{1}{2} \Sigma_k^{-1} \left( \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top \right) \Sigma_k^{-1}, \tag{11.36d}
$$

where $ \sum_{n=1}^N r_{nk} = N_k $.

Setting the derivative to zero for optimality:

$$
-\frac{1}{2} \Sigma_k^{-1} N_k + \frac{1}{2} \Sigma_k^{-1} \left( \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top \right) \Sigma_k^{-1} = 0, \tag{11.37a}
$$

which simplifies to:

$$
N_k I = \left( \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top \right) \Sigma_k^{-1}. \tag{11.37b}
$$

Solving for $ \Sigma_k $:

$$
\Sigma_k^{\text{new}} = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top, \tag{11.38}
$$

which matches Equation (11.30), proving Theorem 11.2. Here, $ r_k $ is the normalized probability vector defined in Equation (11.25).

### Interpretation of the Covariance Update

Similar to the mean update in Equation (11.20), the covariance update in Equation (11.30) can be interpreted as an **importance-weighted expected value** of the squared deviations from the mean. Define the centered data as $ \tilde{X}_k := \{ x_1 - \mu_k, \ldots, x_N - \mu_k \} $. Then $ \Sigma_k^{\text{new}} $ is the weighted covariance of $ \tilde{X}_k $, with weights given by the responsibilities $ r_{nk} $.

These updates reflect the spread of the data points around the updated means, weighted by the responsibilities.

### Monte Carlo Interpretation of Covariance Update

Similar to the update of the mean parameters in Equation (11.20), the covariance update in Equation (11.30) can be interpreted as a **Monte Carlo estimate** of the weighted covariance of the data points $ x_n $ associated with the $ k $-th mixture component:

$$
\Sigma_k^{\text{new}} = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top, \tag{11.30}
$$

Here, the weights are the responsibilities $ r_{nk} $, which represent the probability that the $ n $-th data point belongs to the $ k $-th component. This update estimates the covariance by considering the spread of the data points around the mean $ \mu_k $, with each data point’s contribution weighted by its responsibility.

### Challenges in Obtaining a Closed-Form Solution

As with the updates of the mean parameters, the covariance update depends on all GMM parameters $ \pi_j $, $ \mu_j $, and $ \Sigma_j $ ($ j = 1, \ldots, K $) through the responsibilities $ r_{nk} $ (Equation 11.17). This interdependence prohibits a closed-form solution for the maximum likelihood estimation, necessitating an iterative approach like the Expectation-Maximization (EM) algorithm to optimize the parameters.

```python
import math
import matplotlib.pyplot as plt

class GaussianMixtureModel:
    """
    Gaussian Mixture Model (GMM) implementation for 1D data.
    """
    
    def __init__(self, n_components, means, variances, weights):
        """
        Initialize GMM with given parameters.
        
        Parameters:
        -----------
        n_components : int
            Number of Gaussian components (K)
        means : list
            Means (mu_k)
        variances : list
            Variances (sigma_k^2)
        weights : list
            Mixture weights (pi_k)
        """
        self.K = n_components
        self.mu = means
        self.sigma2 = variances
        self.pi = weights
    
    def gaussian(self, x, mu, sigma2):
        """Compute Gaussian density N(x | mu, sigma2)."""
        coeff = 1.0 / math.sqrt(2 * math.pi * sigma2)
        exp_term = math.exp(-0.5 * (x - mu) ** 2 / sigma2)
        return coeff * exp_term
    
    def compute_responsibilities(self, X):
        """
        Compute responsibilities r_nk (Equation 11.17).
        
        Parameters:
        -----------
        X : list of floats
            1D data
            
        Returns:
        --------
        responsibilities : list of lists
            Matrix of r_nk, shape (N, K)
        """
        n_samples = len(X)
        responsibilities = [[0.0 for _ in range(self.K)] for _ in range(n_samples)]
        for n in range(n_samples):
            total = 0.0
            for k in range(self.K):
                responsibilities[n][k] = self.pi[k] * self.gaussian(X[n], self.mu[k], self.sigma2[k])
                total += responsibilities[n][k]
            for k in range(self.K):
                responsibilities[n][k] /= total if total > 0 else 1.0
        return responsibilities
    
    def update_variances(self, X, responsibilities):
        """
        Update variances using responsibilities (Equation 11.30).
        Since data is 1D, Sigma_k = sigma_k^2.
        
        Parameters:
        -----------
        X : list of floats
            1D data
        responsibilities : list of lists
            Matrix of r_nk, shape (N, K)
            
        Returns:
        --------
        new_variances : list
            Updated variances (sigma_k^2)
        """
        n_samples = len(X)
        Nk = [0.0] * self.K
        new_variances = [0.0] * self.K
        
        # Compute Nk and variance for each component
        for k in range(self.K):
            for n in range(n_samples):
                Nk[k] += responsibilities[n][k]
                new_variances[k] += responsibilities[n][k] * (X[n] - self.mu[k]) ** 2
            new_variances[k] /= Nk[k] if Nk[k] > 0 else 1.0
            new_variances[k] = max(new_variances[k], 1e-6)  # Prevent zero variance
        
        print(f"N_k values: {[round(nk, 3) for nk in Nk]}")
        return new_variances
    
    def predict_density(self, X):
        """
        Compute GMM density at points X.
        
        Parameters:
        -----------
        X : list of floats
            Points to evaluate density at
            
        Returns:
        --------
        densities : list of floats
            GMM density values
        """
        densities = []
        for x in X:
            px = 0.0
            for k in range(self.K):
                px += self.pi[k] * self.gaussian(x, self.mu[k], self.sigma2[k])
            densities.append(px)
        return densities

def plot_gmm_densities(X, gmm_before, gmm_after):
    """Plot GMM densities before and after variance updates."""
    # Create points for density plot
    x_range = [min(X) - 2, max(X) + 2]
    x_points = [x_range[0] + i * (x_range[1] - x_range[0]) / 100 for i in range(101)]
    
    # Compute densities before and after
    densities_before = gmm_before.predict_density(x_points)
    densities_after = gmm_after.predict_density(x_points)
    
    # Plot: two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # (a) Before update
    ax1.scatter(X, [0] * len(X), color='black', marker='o', label='Data Points', s=100)
    ax1.plot(x_points, densities_before, 'k-', label='GMM Density')
    for k in range(gmm_before.K):
        component = [gmm_before.pi[k] * gmm_before.gaussian(x, gmm_before.mu[k], gmm_before.sigma2[k]) for x in x_points]
        ax1.plot(x_points, component, 'g--', label=f'Component {k+1}')
    ax1.set_title('Before Variance Update')
    ax1.set_xlabel('x')
    ax1.set_ylabel('p(x)')
    ax1.set_ylim(bottom=0, top=0.3)
    ax1.legend()
    ax1.grid(True)
    
    # (b) After update
    ax2.scatter(X, [0] * len(X), color='black', marker='o', label='Data Points', s=100)
    ax2.plot(x_points, densities_after, 'k-', label='GMM Density')
    for k in range(gmm_after.K):
        component = [gmm_after.pi[k] * gmm_after.gaussian(x, gmm_after.mu[k], gmm_after.sigma2[k]) for x in x_points]
        ax2.plot(x_points, component, 'g--', label=f'Component {k+1}')
    ax2.set_title('After Variance Update')
    ax2.set_xlabel('x')
    ax2.set_ylabel('p(x)')
    ax2.set_ylim(bottom=0, top=0.3)
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("GMM Variance Update Implementation in Core Python")
    print("=" * 60)
    
    # Dataset
    X = [-3, -2.5, -1, 0, 2, 4, 5]
    
    # Parameters after mean update (Example 11.3)
    means = [-2.7, -0.4, 3.7]
    initial_variances = [1, 0.2, 3]
    weights = [1/3, 1/3, 1/3]
    
    # Create GMM before variance update
    gmm_before = GaussianMixtureModel(
        n_components=3,
        means=means,
        variances=initial_variances,
        weights=weights
    )
    
    # Compute responsibilities (Equation 11.17)
    responsibilities = gmm_before.compute_responsibilities(X)
    
    # Print responsibilities to verify with Equation (11.19)
    print("Responsibilities (r_nk):")
    for row in responsibilities:
        print([round(r, 3) for r in row])
    
    # Update variances (Equation 11.30)
    new_variances = gmm_before.update_variances(X, responsibilities)
    print(f"Initial Variances: {initial_variances}")
    print(f"Updated Variances: {[round(v, 2) for v in new_variances]}")
    
    # Create GMM after variance update
    gmm_after = GaussianMixtureModel(
        n_components=3,
        means=means,
        variances=new_variances,
        weights=weights
    )
    
    # Plot results
    plot_gmm_densities(X, gmm_before, gmm_after)
    
    print("\n=== Demonstration completed ===")
```

## Updating the Mixture Weights

### Theorem 3: Update of the GMM Mixture Weights

The mixture weights of the GMM are updated as:

$$
\pi_k^{\text{new}} = \frac{N_k}{N}, \quad k = 1, \ldots, K, \tag{11.42}
$$

where $ N $ is the number of data points, and $ N_k $ is the total responsibility of the $ k $-th component, defined in Equation (11.24):

$$
N_k = \sum_{n=1}^N r_{nk}.
$$

### Proof of Theorem 3

To derive the update rule for the mixture weights $ \pi_k $ ($ k = 1, \ldots, K $), we must account for the constraint $ \sum_{k=1}^K \pi_k = 1 $. We use Lagrange multipliers (as discussed in Section 7.2) to incorporate this constraint. The Lagrangian is defined as:

$$
\mathcal{L} = L + \lambda \left( \sum_{k=1}^K \pi_k - 1 \right), \tag{11.43a}
$$

where $ L $ is the log-likelihood from Equation (11.10):

$$
\mathcal{L} = \sum_{n=1}^N \log \left( \sum_{k=1}^K \pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k) \right) + \lambda \left( \sum_{k=1}^K \pi_k - 1 \right). \tag{11.43b}
$$

The second term enforces the constraint that the mixture weights sum to 1. We compute the partial derivatives of the Lagrangian with respect to $ \pi_k $ and $ \lambda $.

#### Partial Derivative with Respect to $ \pi_k $

$$
\frac{\partial \mathcal{L}}{\partial \pi_k} = \sum_{n=1}^N \frac{\mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)} + \lambda, \tag{11.44a}
$$

Recognizing that the fraction is the responsibility $ r_{nk} $ (Equation 11.17), we simplify:

$$
\frac{\partial \mathcal{L}}{\partial \pi_k} = \sum_{n=1}^N \frac{\pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)} \cdot \frac{1}{\pi_k} + \lambda = \frac{1}{\pi_k} \sum_{n=1}^N r_{nk} + \lambda = \frac{N_k}{\pi_k} + \lambda, \tag{11.44b}
$$

where $ N_k = \sum_{n=1}^N r_{nk} $.

#### Partial Derivative with Respect to $ \lambda $

$$
\frac{\partial \mathcal{L}}{\partial \lambda} = \sum_{k=1}^K \pi_k - 1. \tag{11.45}
$$

Setting both partial derivatives to zero (necessary condition for an optimum), we obtain the system of equations:

$$
\frac{N_k}{\pi_k} + \lambda = 0 \quad \Rightarrow \quad \pi_k = -\frac{N_k}{\lambda}, \tag{11.46}
$$

$$
\sum_{k=1}^K \pi_k = 1. \tag{11.47}
$$

Substitute Equation (11.46) into Equation (11.47):

$$
\sum_{k=1}^K \pi_k = \sum_{k=1}^K \left(-\frac{N_k}{\lambda}\right) = -\frac{1}{\lambda} \sum_{k=1}^K N_k = 1 \quad \Rightarrow \quad -\frac{1}{\lambda} \sum_{k=1}^K N_k = 1.
$$

Since $ \sum_{k=1}^K N_k = \sum_{k=1}^K \sum_{n=1}^N r_{nk} = N $ (because the responsibilities sum to 1 for each data point), we have:

$$
-\frac{N}{\lambda} = 1 \quad \Rightarrow \quad \lambda = -N. \tag{11.48}
$$

Substitute $ \lambda = -N $ back into Equation (11.46):

$$
\pi_k = -\frac{N_k}{\lambda} = -\frac{N_k}{-N} = \frac{N_k}{N}, \tag{11.49}
$$

which matches Equation (11.42), proving Theorem 11.3.

### Interpretation of Mixture Weight Update

The updated mixture weight $ \pi_k^{\text{new}} = \frac{N_k}{N} $ represents the ratio of the total responsibility of the $ k $-th component to the number of data points. Since $ N = \sum_{k=1}^K N_k $, $ N $ is the total responsibility across all components. Thus, $ \pi_k $ reflects the relative importance of the $ k $-th mixture component for the dataset.

### Remark: Dependency on Responsibilities

Since $ N_k = \sum_{n=1}^N r_{nk} $, the update equation for the mixture weights $ \pi_k $ (Equation 11.42) depends on all parameters $ \pi_j $, $ \mu_j $, and $ \Sigma_j $ ($ j = 1, \ldots, K $) via the responsibilities $ r_{nk} $ (Equation 11.17). This interdependence prevents a closed-form solution, as noted previously. $ \diamond $

Compared to the initial GMM in Figure 11.3 (with means $ \mu_1 = -4 $, $ \mu_2 = 0 $, $ \mu_3 = 8 $; variances $ \sigma_1^2 = 1 $, $ \sigma_2^2 = 0.2 $, $ \sigma_3^2 = 3 $; and weights $ \pi_k = \frac{1}{3} $), the updated GMM has shifted its density mass closer to the data points. The fit in Figure 11.7(b) is significantly better than the initialization, as evidenced by the log-likelihood, which increased from $ -28.3 $ (initialization) to $ -14.4 $ after one full update cycle.

```python
import math
import matplotlib.pyplot as plt

class GaussianMixtureModel:
    """
    Gaussian Mixture Model (GMM) implementation for 1D data.
    """
    
    def __init__(self, n_components, means, variances, weights):
        """
        Initialize GMM with given parameters.
        
        Parameters:
        -----------
        n_components : int
            Number of Gaussian components (K)
        means : list
            Means (mu_k)
        variances : list
            Variances (sigma_k^2)
        weights : list
            Mixture weights (pi_k)
        """
        self.K = n_components
        self.mu = means
        self.sigma2 = variances
        self.pi = weights
    
    def gaussian(self, x, mu, sigma2):
        """Compute Gaussian density N(x | mu, sigma2)."""
        coeff = 1.0 / math.sqrt(2 * math.pi * sigma2)
        exp_term = math.exp(-0.5 * (x - mu) ** 2 / sigma2)
        return coeff * exp_term
    
    def compute_responsibilities(self, X):
        """
        Compute responsibilities r_nk (Equation 11.17).
        
        Parameters:
        -----------
        X : list of floats
            1D data
            
        Returns:
        --------
        responsibilities : list of lists
            Matrix of r_nk, shape (N, K)
        """
        n_samples = len(X)
        responsibilities = [[0.0 for _ in range(self.K)] for _ in range(n_samples)]
        for n in range(n_samples):
            total = 0.0
            for k in range(self.K):
                responsibilities[n][k] = self.pi[k] * self.gaussian(X[n], self.mu[k], self.sigma2[k])
                total += responsibilities[n][k]
            for k in range(self.K):
                responsibilities[n][k] /= total if total > 0 else 1.0
        return responsibilities
    
    def compute_log_likelihood(self, X):
        """
        Compute the log-likelihood (Equation 11.10).
        
        Parameters:
        -----------
        X : list of floats
            1D data
            
        Returns:
        --------
        log_likelihood : float
            Log-likelihood value
        """
        n_samples = len(X)
        log_likelihood = 0.0
        for n in range(n_samples):
            px = 0.0
            for k in range(self.K):
                px += self.pi[k] * self.gaussian(X[n], self.mu[k], self.sigma2[k])
            log_likelihood += math.log(px) if px > 0 else -1e10
        return log_likelihood
    
    def update_weights(self, X, responsibilities):
        """
        Update mixture weights using responsibilities (Equation 11.42).
        
        Parameters:
        -----------
        X : list of floats
            1D data
        responsibilities : list of lists
            Matrix of r_nk, shape (N, K)
            
        Returns:
        --------
        new_weights : list
            Updated mixture weights (pi_k)
        """
        n_samples = len(X)
        Nk = [0.0] * self.K
        
        # Compute Nk for each component
        for k in range(self.K):
            for n in range(n_samples):
                Nk[k] += responsibilities[n][k]
        
        # Update weights: pi_k = Nk / N
        new_weights = [nk / n_samples for nk in Nk]
        
        print(f"N_k values: {[round(nk, 3) for nk in Nk]}")
        return new_weights
    
    def predict_density(self, X):
        """
        Compute GMM density at points X.
        
        Parameters:
        -----------
        X : list of floats
            Points to evaluate density at
            
        Returns:
        --------
        densities : list of floats
            GMM density values
        """
        densities = []
        for x in X:
            px = 0.0
            for k in range(self.K):
                px += self.pi[k] * self.gaussian(x, self.mu[k], self.sigma2[k])
            densities.append(px)
        return densities

def plot_gmm_densities(X, gmm_before, gmm_after):
    """Plot GMM densities before and after weight updates, mimicking Figure 11.7."""
    # Create points for density plot
    x_range = [min(X) - 2, max(X) + 2]
    x_points = [x_range[0] + i * (x_range[1] - x_range[0]) / 100 for i in range(101)]
    
    # Compute densities before and after
    densities_before = gmm_before.predict_density(x_points)
    densities_after = gmm_after.predict_density(x_points)
    
    # Plot: two subplots like Figure 11.7
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # (a) Before update
    ax1.scatter(X, [0] * len(X), color='black', marker='o', label='Data Points', s=100)
    ax1.plot(x_points, densities_before, 'k-', label='GMM Density')
    for k in range(gmm_before.K):
        component = [gmm_before.pi[k] * gmm_before.gaussian(x, gmm_before.mu[k], gmm_before.sigma2[k]) for x in x_points]
        ax1.plot(x_points, component, 'g--', label=f'Component {k+1}')
    ax1.set_title('(a) Before Weight Update')
    ax1.set_xlabel('x')
    ax1.set_ylabel('p(x)')
    ax1.set_ylim(bottom=0, top=0.35)
    ax1.legend()
    ax1.grid(True)
    
    # (b) After update
    ax2.scatter(X, [0] * len(X), color='black', marker='o', label='Data Points', s=100)
    ax2.plot(x_points, densities_after, 'k-', label='GMM Density')
    for k in range(gmm_after.K):
        component = [gmm_after.pi[k] * gmm_after.gaussian(x, gmm_after.mu[k], gmm_after.sigma2[k]) for x in x_points]
        ax2.plot(x_points, component, 'g--', label=f'Component {k+1}')
    ax2.set_title('(b) After Weight Update')
    ax2.set_xlabel('x')
    ax2.set_ylabel('p(x)')
    ax2.set_ylim(bottom=0, top=0.35)
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("GMM Weight Update Implementation in Core Python")
    print("=" * 60)
    
    # Dataset
    X = [-3, -2.5, -1, 0, 2, 4, 5]
    
    # Initial GMM for log-likelihood computation
    initial_means = [-4, 0, 8]
    initial_variances = [1, 0.2, 3]
    initial_weights = [1/3, 1/3, 1/3]
    gmm_initial = GaussianMixtureModel(
        n_components=3,
        means=initial_means,
        variances=initial_variances,
        weights=initial_weights
    )
    initial_log_likelihood = gmm_initial.compute_log_likelihood(X)
    print(f"Initial Log-Likelihood: {round(initial_log_likelihood, 1)}")
    
    # Parameters after mean and variance updates (Examples 11.3 and 11.4)
    means = [-2.7, -0.4, 3.7]
    variances = [0.14, 0.44, 1.53]
    weights = [1/3, 1/3, 1/3]
    
    # Create GMM before weight update
    gmm_before = GaussianMixtureModel(
        n_components=3,
        means=means,
        variances=variances,
        weights=weights
    )
    
    # Compute responsibilities (Equation 11.17)
    responsibilities = gmm_before.compute_responsibilities(X)
    
    # Print responsibilities to verify with Equation (11.19)
    print("Responsibilities (r_nk):")
    for row in responsibilities:
        print([round(r, 3) for r in row])
    
    # Update weights (Equation 11.42)
    new_weights = gmm_before.update_weights(X, responsibilities)
    print(f"Initial Weights: {[round(w, 2) for w in weights]}")
    print(f"Updated Weights: {[round(w, 2) for w in new_weights]}")
    
    # Create GMM after weight update
    gmm_after = GaussianMixtureModel(
        n_components=3,
        means=means,
        variances=variances,
        weights=new_weights
    )
    
    # Compute final log-likelihood
    final_log_likelihood = gmm_after.compute_log_likelihood(X)
    print(f"Final Log-Likelihood: {round(final_log_likelihood, 1)}")
    
    # Plot results
    plot_gmm_densities(X, gmm_before, gmm_after)
    
    print("\n=== Demonstration completed ===")
```

##  EM Algorithm

The updates for the GMM parameters $ \mu_k $, $ \Sigma_k $, and $ \pi_k $, given by Equations (11.20), (11.30), and (11.42), do not provide a closed-form solution because the responsibilities $ r_{nk} $ depend on these parameters in a complex way:

- Means: $ \mu_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} x_n $, \tag{11.20}
- Covariances: $ \Sigma_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top $, \tag{11.30}
- Mixture Weights: $ \pi_k = \frac{N_k}{N} $. \tag{11.42}

However, these updates suggest a simple iterative scheme for parameter estimation via maximum likelihood. The **Expectation-Maximization (EM) algorithm**, proposed by Dempster et al. (1977), is a general iterative method for learning parameters (via maximum likelihood or MAP) in mixture models and, more broadly, latent-variable models.

### EM Algorithm Overview

In the context of a Gaussian Mixture Model, the EM algorithm proceeds as follows:
1. Choose initial values for $ \mu_k $, $ \Sigma_k $, and $ \pi_k $.
2. Alternate between the following steps until convergence:
   - **E-step**: Compute the responsibilities $ r_{nk} $, which represent the posterior probability that data point $ x_n $ belongs to the $ k $-th mixture component.
   - **M-step**: Use the updated responsibilities to reestimate the parameters $ \mu_k $, $ \Sigma_k $, and $ \pi_k $.

Each iteration of the EM algorithm is guaranteed to increase the log-likelihood function (Neal and Hinton, 1999). Convergence can be assessed by monitoring the log-likelihood or the parameter values directly.

### EM Algorithm for GMMs

A concrete instantiation of the EM algorithm for estimating GMM parameters is as follows:

1. **Initialize** $ \mu_k $, $ \Sigma_k $, and $ \pi_k $.
2. **E-step**: Compute the responsibilities $ r_{nk} $ for each data point $ x_n $ using the current parameters:

$$
r_{nk} = \frac{\pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)}. \tag{11.53}
$$

3. **M-step**: Reestimate the parameters using the current responsibilities:
   - Update the means:

$$
\mu_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} x_n, \tag{11.54}
$$

   where $ N_k = \sum_{n=1}^N r_{nk} $.
   - Update the covariances, using the updated means:

$$
\Sigma_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top. \tag{11.55}
$$

   - Update the mixture weights:

$$
\pi_k = \frac{N_k}{N}. \tag{11.56}
$$

4. Repeat steps 2 and 3 until convergence.

### Example 6: GMM Fit

In our running example (dataset $ X = \{-3, -2.5, -1, 0, 2, 4, 5\} $, $ K = 3 $), we apply the EM algorithm. After five iterations, the EM algorithm converges, yielding the final GMM fit shown in Fig.8:


- **Fig.8(a)**: Displays the final GMM density and its individual components after convergence.
- **Fig.8(b)**: Plots the negative log-likelihood as a function of the EM iterations, decreasing from an initial value (e.g., 28.3) to a final value (e.g., 14.4 after the first iteration, continuing to decrease over five iterations).

This demonstrates the EM algorithm’s ability to iteratively improve the GMM fit by maximizing the log-likelihood.

```python
import math
import random

# Normal probability density function
def gaussian(x, mu, sigma2):
    coeff = 1.0 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

# Initialize data and parameters
X = [-3, -2.5, -1, 0, 2, 4, 5]
K = 3
N = len(X)

# Initialize means, variances, and weights
mus = [-2, 0, 3]
sigmas2 = [1, 1, 1]
pis = [1.0 / K for _ in range(K)]

# Responsibilities: r[n][k]
r = [[0.0 for _ in range(K)] for _ in range(N)]

# EM iterations
max_iter = 10
for iteration in range(max_iter):
    # E-step
    for n in range(N):
        total = 0.0
        for k in range(K):
            r[n][k] = pis[k] * gaussian(X[n], mus[k], sigmas2[k])
            total += r[n][k]
        for k in range(K):
            r[n][k] /= total if total > 0 else 1e-12  # Normalize

    # M-step
    for k in range(K):
        N_k = sum(r[n][k] for n in range(N))
        # Update mean
        mus[k] = sum(r[n][k] * X[n] for n in range(N)) / N_k
        # Update variance
        sigmas2[k] = sum(r[n][k] * (X[n] - mus[k])**2 for n in range(N)) / N_k
        # Update mixing weight
        pis[k] = N_k / N

    # Optionally print diagnostics
    print(f"Iteration {iteration + 1}")
    for k in range(K):
        print(f"  Component {k + 1}: mean = {mus[k]:.4f}, var = {sigmas2[k]:.4f}, pi = {pis[k]:.4f}")
```

# EM Algorithm for Gaussian Mixture Models (GMM)

We fit a Gaussian Mixture Model (GMM) with \( K = 3 \) components to a two-dimensional dataset using the Expectation-Maximization (EM) algorithm.

---

## EM Algorithm Overview

The EM algorithm iteratively improves the GMM fit via:

- **E-step**: Estimate responsibilities $ r_{nk} $
- **M-step**: Update parameters $ \mu_k, \Sigma_k, \pi_k $

---

## Parameter Update Equations

- Mean:

$$
\mu_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} x_n
$$

- Covariance:

$$
\Sigma_k = \frac{1}{N_k} \sum_{n=1}^N r_{nk} (x_n - \mu_k)(x_n - \mu_k)^\top
$$

- Mixture weight:

$$
\pi_k = \frac{N_k}{N}
$$

where $ N_k = \sum_{n=1}^N r_{nk} $

---

## Responsibility Computation

The responsibilities are given by:

$$
r_{nk} = \frac{\pi_k \mathcal{N}(x_n \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_n \mid \mu_j, \Sigma_j)}
$$

---

## Example GMM Fit

Given dataset:

$$
X = \{-3, -2.5, -1, 0, 2, 4, 5\}, \quad K = 3
$$

$$
p(x) = 0.29 \mathcal{N}(x \mid -2.75, 0.06) + 0.28 \mathcal{N}(x \mid -0.50, 0.25) + 0.43 \mathcal{N}(x \mid 3.64, 1.63)
$$

```python
import math
import random
import matplotlib.pyplot as plt

# Generate synthetic 2D dataset
X = [[-6, -5], [-5.5, -4.5], [-6.5, -4], [0.5, 0.5], [1.5, 1.2], [1, 2], [3, 4], [4, 4], [4.5, 4.2], [5, 5]]

N = len(X)
K = 3  # Number of components
d = 2  # Dimension

# Initialize means (μ_k), diagonal variances (Σ_k), and priors (π_k)
mus = [[-5, -5], [1, 1], [4.5, 4]]
sigmas = [[1, 1] for _ in range(K)]
pis = [1 / K for _ in range(K)]

# Initialize responsibilities r[n][k]
r = [[0] * K for _ in range(N)]

def gaussian_2d(x, mu, sigma):
    """Multivariate Gaussian density with diagonal covariance"""
    det_sigma = sigma[0] * sigma[1]
    norm_const = 1 / (2 * math.pi * math.sqrt(det_sigma))
    exp_val = -0.5 * sum(((x[i] - mu[i]) ** 2) / sigma[i] for i in range(2))
    return norm_const * math.exp(exp_val)

log_likelihoods = []

for iteration in range(30):
    # E-step
    for n in range(N):
        total = 0
        for k in range(K):
            r[n][k] = pis[k] * gaussian_2d(X[n], mus[k], sigmas[k])
            total += r[n][k]
        for k in range(K):
            r[n][k] /= total if total > 0 else 1e-12

    # M-step
    for k in range(K):
        N_k = sum(r[n][k] for n in range(N))
        # Update mean
        mus[k] = [sum(r[n][k] * X[n][dim] for n in range(N)) / N_k for dim in range(d)]
        # Update variance (diagonal only)
        sigmas[k] = [
            sum(r[n][k] * (X[n][dim] - mus[k][dim]) ** 2 for n in range(N)) / N_k
            for dim in range(d)
        ]
        # Update mixing weights
        pis[k] = N_k / N

    # Compute log-likelihood
    ll = 0
    for n in range(N):
        prob = sum(pis[k] * gaussian_2d(X[n], mus[k], sigmas[k]) for k in range(K))
        ll += math.log(prob + 1e-12)
    log_likelihoods.append(ll)

# Plotting: Data and GMM component means
colors = ['red', 'green', 'blue']
plt.figure(figsize=(12, 5))

# (a) Dataset with cluster means
plt.subplot(1, 2, 1)
for k in range(K):
    cluster_points = [X[n] for n in range(N) if max(r[n]) == r[n][k]]
    xs, ys = zip(*cluster_points) if cluster_points else ([], [])
    plt.scatter(xs, ys, color=colors[k], label=f'Component {k+1}', alpha=0.6)
    plt.scatter(mus[k][0], mus[k][1], color=colors[k], edgecolor='black', s=100, marker='X')

plt.title("2D GMM Fit (Means shown as X)")
plt.xlabel("x₁")
plt.ylabel("x₂")
plt.legend()

# (b) Log-Likelihood over EM iterations
plt.subplot(1, 2, 2)
plt.plot(range(1, len(log_likelihoods)+1), log_likelihoods, marker='o')
plt.title("Log-Likelihood vs EM Iteration")
plt.xlabel("EM Iteration")
plt.ylabel("Log-Likelihood")

plt.tight_layout()
plt.show()
```
