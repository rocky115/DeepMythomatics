## Product of Gaussian Densities

For linear regression (Chapter 9), we need to compute a Gaussian likelihood. Furthermore, we may wish to assume a Gaussian prior (Section 9.3). We apply Bayes’ Theorem to compute the posterior, which results in a multiplication of the likelihood and the prior, that is, the multiplication of two Gaussian densities. The product of two Gaussians $\mathcal{N}(x \mid a, A) \mathcal{N}(x \mid b, B)$ is a Gaussian distribution scaled by a $c \in \mathbb{R}$, given by $c \mathcal{N}(x \mid c, C)$ with

$$
C = \left(A^{-1} + B^{-1}\right)^{-1} \tag{6.74}
$$

$$
c = C \left(A^{-1} a + B^{-1} b\right) \tag{6.75}
$$

$$
c = (2\pi)^{-\frac{D}{2}} |A + B|^{-\frac{1}{2}} \exp\left(-\frac{1}{2}(a - b)^\top (A + B)^{-1} (a - b)\right). \tag{6.76}
$$

The scaling constant $c$ itself can be written in the form of a Gaussian density either in $a$ or in $b$ with an “inflated” covariance matrix $A + B$, i.e., $c = \mathcal{N}(a \mid b, A + B) = \mathcal{N}(b \mid a, A + B)$.

### Remark

For notation convenience, we will sometimes use $\mathcal{N}(x \mid m, S)$ to describe the functional form of a Gaussian density even if $x$ is not a random variable. We have just done this in the preceding demonstration when we wrote

$$
c = \mathcal{N}(a \mid b, A + B) = \mathcal{N}(b \mid a, A + B). \tag{6.77}
$$

Here, neither $a$ nor $b$ are random variables. However, writing $c$ in this way is more compact than (6.76). $\diamond$

## 6.5.3 Sums and Linear Transformations

If $X$ and $Y$ are independent Gaussian random variables (i.e., the joint distribution is given as $p(x, y) = p(x)p(y)$) with $p(x) = \mathcal{N}(x \mid \mu_x, \Sigma_x)$ and $p(y) = \mathcal{N}(y \mid \mu_y, \Sigma_y)$, then $x + y$ is also Gaussian distributed and given by

$$
p(x + y) = \mathcal{N}(\mu_x + \mu_y, \Sigma_x + \Sigma_y). \tag{6.78}
$$

Knowing that $p(x + y)$ is Gaussian, the mean and covariance matrix can be determined immediately using the results from (6.46) through (6.49). This property will be important when we consider i.i.d. Gaussian noise acting on random variables, as is the case for linear regression (Chapter 9).

### Example 6.7

Since expectations are linear operations, we can obtain the weighted sum of independent Gaussian random variables

$$
p(ax + by) = \mathcal{N}(a\mu_x + b\mu_y, a^2 \Sigma_x + b^2 \Sigma_y). \tag{6.79}
$$

### Remark

A case that will be useful in Chapter 11 is the weighted sum of Gaussian *densities*. This is different from the weighted sum of Gaussian random variables. $\diamond$

In Theorem 6.12, the random variable $x$ is from a density that is a mixture of two densities $p_1(x)$ and $p_2(x)$, weighted by $\alpha$. The theorem can be generalized to the multivariate random variable case, since linearity of expectations holds also for multivariate random variables. However, the idea of a squared random variable needs to be replaced by $xx^\top$.

### Theorem 6.12

Consider a mixture of two univariate Gaussian densities

$$
p(x) = \alpha p_1(x) + (1 - \alpha) p_2(x), \tag{6.80}
$$

where the scalar $0 < \alpha < 1$ is the mixture weight, and $p_1(x)$ and $p_2(x)$ are univariate Gaussian densities (Equation (6.62)) with different parameters, i.e., $(\mu_1, \sigma_1^2) \neq (\mu_2, \sigma_2^2)$. Then the mean of the mixture density $p(x)$ is given by the weighted sum of the means of each random variable:

$$
\mathbb{E}[x] = \alpha \mu_1 + (1 - \alpha) \mu_2. \tag{6.81}
$$

The variance of the mixture density $p(x)$ is given by

$$
\text{V}[x] = \alpha \sigma_1^2 + (1 - \alpha) \sigma_2^2 + \left(\alpha \mu_1^2 + (1 - \alpha) \mu_2^2\right) - \left[\alpha \mu_1 + (1 - \alpha) \mu_2\right]^2. \tag{6.82}
$$

### Proof

The mean of the mixture density $p(x)$ is given by the weighted sum of the means of each random variable. We apply the definition of the mean (Definition 6.4), and plug in our mixture (6.80), which yields

$$
\mathbb{E}[x] = \int_{-\infty}^{\infty} x p(x) \, dx \tag{6.83a}
$$

$$
= \int_{-\infty}^{\infty} \left(\alpha x p_1(x) + (1 - \alpha) x p_2(x)\right) \, dx \tag{6.83b}
$$

$$
= \alpha \int_{-\infty}^{\infty} x p_1(x) \, dx + (1 - \alpha) \int_{-\infty}^{\infty} x p_2(x) \, dx \tag{6.83c}
$$

$$
= \alpha \mu_1 + (1 - \alpha) \mu_2. \tag{6.83d}
$$

To compute the variance, we can use the raw-score version of the variance from (6.44), which requires an expression of the expectation of the squared random variable. Here we use the definition of an expectation of a function (the square) of a random variable (Definition 6.3),

$$
\mathbb{E}[x^2] = \int_{-\infty}^{\infty} x^2 p(x) \, dx \tag{6.84a}
$$

$$
= \int_{-\infty}^{\infty} \left(\alpha x^2 p_1(x) + (1 - \alpha) x^2 p_2(x)\right) \, dx \tag{6.84b}
$$

```python
import random
import math

# --- Gaussian Distribution Functions ---
def univariate_gaussian_pdf(x, mu, sigma2):
    """Compute the univariate Gaussian PDF (Equation 6.62)."""
    coeff = 1 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

# --- Product of Gaussian Densities (Section 6.5.2) ---
def product_of_gaussians(a, A, b, B):
    """
    Compute the product of two univariate Gaussian densities (Equations 6.74-6.76).
    a, b: means of the Gaussians
    A, B: variances of the Gaussians
    Returns: (c, C, scaling_constant) where the product is scaling_constant * N(x|c, C)
    """
    # Equation 6.74: C = (A^-1 + B^-1)^-1
    A_inv = 1 / A
    B_inv = 1 / B
    C = 1 / (A_inv + B_inv)

    # Equation 6.75: c = C (A^-1 a + B^-1 b)
    c = C * (A_inv * a + B_inv * b)

    # Equation 6.76: scaling constant (simplified for D=1)
    scaling_constant = (2 * math.pi) ** (-0.5) * (A + B) ** (-0.5) * math.exp(-0.5 * (a - b) ** 2 / (A + B))

    return c, C, scaling_constant

# --- Generate Gaussian Samples ---
def generate_gaussian_samples(mu, sigma2, n_samples=1000):
    """Generate samples from a univariate Gaussian using Box-Muller transform."""
    samples = [random.gauss(mu, math.sqrt(sigma2)) for _ in range(n_samples)]
    return samples

# --- Compute Mean and Variance from Samples ---
def compute_mean(samples):
    return sum(samples) / len(samples)

def compute_variance(samples):
    mean = compute_mean(samples)
    return sum((x - mean) ** 2 for x in samples) / len(samples)

# --- Mixture of Gaussian Densities ---
def mixture_gaussian_pdf(x, alpha, mu1, sigma1_2, mu2, sigma2_2):
    """Compute the PDF of a mixture of two Gaussian densities (Equation 6.80)."""
    p1 = univariate_gaussian_pdf(x, mu1, sigma1_2)
    p2 = univariate_gaussian_pdf(x, mu2, sigma2_2)
    return alpha * p1 + (1 - alpha) * p2

# --- Demonstration ---
def demonstrate_gaussian_properties():
    """
    Demonstrate properties of Gaussian distributions (Sections 6.5.2, 6.5.3).
    - Compute the product of two Gaussian densities (Equations 6.74-6.76)
    - Verify the sum of independent Gaussians (Equation 6.78)
    - Compute the weighted sum (Equation 6.79)
    - Compute mean and variance of a mixture of Gaussians (Equations 6.81-6.82)
    """
    print("=== Gaussian Distribution Properties ===")
    print("Sections 6.5.2, 6.5.3: Product, Sums, and Mixtures of Gaussians\n")

    # Step 1: Product of Gaussian Densities (Section 6.5.2)
    print("Step 1: Product of Gaussian Densities (Equations 6.74-6.76)")
    a, A = 1.0, 0.5  # First Gaussian: N(x|1, 0.5)
    b, B = 2.0, 1.0  # Second Gaussian: N(x|2, 1.0)
    c, C, scaling_constant = product_of_gaussians(a, A, b, B)
    print(f"First Gaussian: N(x|{a}, {A})")
    print(f"Second Gaussian: N(x|{b}, {B})")
    print(f"Product: {scaling_constant:.4f} * N(x|{c:.4f}, {C:.4f})")
    print()

    # Step 2: Sum of Independent Gaussians (Equation 6.78)
    print("Step 2: Sum of Independent Gaussians (Equation 6.78)")
    mu_x, sigma_x2 = 1.0, 0.5
    mu_y, sigma_y2 = 2.0, 1.0
    X = generate_gaussian_samples(mu_x, sigma_x2)
    Y = generate_gaussian_samples(mu_y, sigma_y2)
    X_plus_Y = [x + y for x, y in zip(X, Y)]
    empirical_mean = compute_mean(X_plus_Y)
    empirical_var = compute_variance(X_plus_Y)
    theoretical_mean = mu_x + mu_y
    theoretical_var = sigma_x2 + sigma_y2
    print(f"Theoretical p(x+y) = N({theoretical_mean}, {theoretical_var})")
    print(f"Empirical mean of x+y: {empirical_mean:.4f}")
    print(f"Empirical variance of x+y: {empirical_var:.4f}")
    print()

    # Step 3: Weighted Sum of Gaussians (Equation 6.79, Example 6.7)
    print("Step 3: Weighted Sum of Gaussians (Equation 6.79)")
    a, b = 2.0, 3.0
    Z = [a * x + b * y for x, y in zip(X, Y)]
    empirical_mean_Z = compute_mean(Z)
    empirical_var_Z = compute_variance(Z)
    theoretical_mean_Z = a * mu_x + b * mu_y
    theoretical_var_Z = (a ** 2) * sigma_x2 + (b ** 2) * sigma_y2
    print(f"Theoretical p({a}x + {b}y) = N({theoretical_mean_Z}, {theoretical_var_Z})")
    print(f"Empirical mean of {a}x + {b}y: {empirical_mean_Z:.4f}")
    print(f"Empirical variance of {a}x + {b}y: {empirical_var_Z:.4f}")
    print()

    # Step 4: Mixture of Gaussian Densities (Theorem 6.12)
    print("Step 4: Mixture of Gaussian Densities (Equations 6.81-6.82)")
    alpha = 0.4
    mu1, sigma1_2 = 0.0, 0.5
    mu2, sigma2_2 = 2.0, 1.0
    # Compute theoretical mean and variance
    mean_mixture = alpha * mu1 + (1 - alpha) * mu2  # Equation 6.81
    term1 = alpha * sigma1_2 + (1 - alpha) * sigma2_2
    term2 = alpha * (mu1 ** 2) + (1 - alpha) * (mu2 ** 2)
    term3 = (alpha * mu1 + (1 - alpha) * mu2) ** 2
    var_mixture = term1 + term2 - term3  # Equation 6.82
    print(f"Mixture: p(x) = {alpha} N(x|{mu1}, {sigma1_2}) + {1-alpha} N(x|{mu2}, {sigma2_2})")
    print(f"Theoretical mean of mixture: {mean_mixture:.4f}")
    print(f"Theoretical variance of mixture: {var_mixture:.4f}")
    print()

# --- Main Execution ---
if __name__ == "__main__":
    print("Gaussian Properties Analysis")
    print("=" * 60)

    # Run demonstration
    demonstrate_gaussian_properties()

    print("\n" + "=" * 60)
    print("Summary of Key Results:")
    print("• Computed product of two Gaussian densities (Section 6.5.2)")
    print("• Verified sum of independent Gaussians (Equation 6.78)")
    print("• Computed weighted sum of Gaussians (Equation 6.79)")
    print("• Computed mean and variance of a mixture of Gaussians (Equations 6.81-6.82)")
```

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # Needed for Seaborn plotting, as Seaborn works with NumPy arrays

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Gaussian Distribution Functions ---
def univariate_gaussian_pdf(x, mu, sigma2):
    """Compute the univariate Gaussian PDF (Equation 6.62)."""
    coeff = 1 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

# --- Generate Gaussian Samples ---
def generate_gaussian_samples(mu, sigma2, n_samples=1000):
    """Generate samples from a univariate Gaussian using Box-Muller transform."""
    samples = [random.gauss(mu, math.sqrt(sigma2)) for _ in range(n_samples)]
    return samples

# --- Mixture of Gaussian Densities ---
def mixture_gaussian_pdf(x, alpha, mu1, sigma1_2, mu2, sigma2_2):
    """Compute the PDF of a mixture of two Gaussian densities (Equation 6.80)."""
    p1 = univariate_gaussian_pdf(x, mu1, sigma1_2)
    p2 = univariate_gaussian_pdf(x, mu2, sigma2_2)
    return alpha * p1 + (1 - alpha) * p2

# --- Generate Bivariate Gaussian Samples ---
def generate_bivariate_gaussian_samples(mu, Sigma, n_samples=1000):
    """
    Generate samples from a bivariate Gaussian using the Box-Muller transform.
    """
    mu_x, mu_y = mu
    Sigma_xx, Sigma_xy = Sigma[0][0], Sigma[0][1]
    Sigma_yx, Sigma_yy = Sigma[1][0], Sigma[1][1]

    # Cholesky decomposition of Sigma (simplified for 2x2 matrix)
    L11 = math.sqrt(Sigma_xx)
    L21 = Sigma_yx / L11
    L22 = math.sqrt(Sigma_yy - (L21 ** 2))

    # Standard normal samples
    Z1 = [random.gauss(0, 1) for _ in range(n_samples)]
    Z2 = [random.gauss(0, 1) for _ in range(n_samples)]

    # Transform to correlated variables: [X, Y] = mu + L * [Z1, Z2]
    X = [mu_x + L11 * z1 for z1 in Z1]
    Y = [mu_y + L21 * z1 + L22 * z2 for z1, z2 in zip(Z1, Z2)]

    return X, Y

# --- Compute Conditional Parameters (Equations 6.70, 6.71) ---
def compute_conditional_parameters(mu, Sigma, y_obs):
    """
    Compute the conditional mean and variance of x1 given x2 = y_obs (Equations 6.66, 6.67).
    """
    mu_x, mu_y = mu
    Sigma_xx, Sigma_xy = Sigma[0][0], Sigma[0][1]
    Sigma_yx, Sigma_yy = Sigma[1][0], Sigma[1][1]

    # Sigma_yy inverse (scalar for univariate y)
    Sigma_yy_inv = 1 / Sigma_yy
    # Compute conditional mean (Equation 6.66)
    mu_x_given_y = mu_x + Sigma_xy * Sigma_yy_inv * (y_obs - mu_y)
    # Compute conditional variance (Equation 6.67)
    Sigma_x_given_y = Sigma_xx - Sigma_xy * Sigma_yy_inv * Sigma_yx

    return mu_x_given_y, Sigma_x_given_y

# --- Plotting Functions ---
def plot_univariate_gaussian():
    """Plot a univariate Gaussian (Figure 6.8a) with samples."""
    mu, sigma2 = 0, 1  # Standard normal
    samples = generate_gaussian_samples(mu, sigma2, n_samples=100)

    # Generate points for the PDF
    x = np.linspace(-5, 5, 100)
    pdf = [univariate_gaussian_pdf(xi, mu, sigma2) for xi in x]

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=pdf, label="Univariate Gaussian p(x)", color="teal")
    sns.histplot(samples, stat="density", color="orange", alpha=0.5, label="Samples (n=100)")
    plt.axvline(mu, color="red", linestyle="--", label="Mean")
    plt.axvline(mu + 2 * math.sqrt(sigma2), color="red", linestyle="--", alpha=0.5, label="Mean ± 2σ")
    plt.axvline(mu - 2 * math.sqrt(sigma2), color="red", linestyle="--", alpha=0.5)
    plt.title("Figure 6.8a: Univariate Gaussian")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

def plot_marginal_and_conditional():
    """Plot marginal and conditional distributions (Figures 6.9b, 6.9c)."""
    mu = [0, 2]
    Sigma = [[0.3, -1], [-1, 5]]
    mu_x1, Sigma_xx = mu[0], Sigma[0][0]  # Marginal parameters
    y_obs = -1
    mu_x1_given_x2, Sigma_x1_given_x2 = compute_conditional_parameters(mu, Sigma, y_obs)

    # Generate points for the PDFs
    x = np.linspace(-1.5, 1.5, 100)
    marginal_pdf = [univariate_gaussian_pdf(xi, mu_x1, Sigma_xx) for xi in x]
    conditional_pdf = [univariate_gaussian_pdf(xi, mu_x1_given_x2, Sigma_x1_given_x2) for xi in x]

    # Plot Marginal (Figure 6.9b)
    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=marginal_pdf, label="Marginal p(x1)", color="red")
    plt.axvline(mu_x1, color="red", linestyle="--", label="Mean")
    plt.axvline(mu_x1 + 2 * math.sqrt(Sigma_xx), color="red", linestyle="--", alpha=0.5, label="Mean ± 2σ")
    plt.axvline(mu_x1 - 2 * math.sqrt(Sigma_xx), color="red", linestyle="--", alpha=0.5)
    plt.title("Figure 6.9b: Marginal Distribution p(x1)")
    plt.xlabel("x1")
    plt.ylabel("p(x1)")
    plt.legend()
    plt.show()

    # Plot Conditional (Figure 6.9c)
    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=conditional_pdf, label="Conditional p(x1|x2=-1)", color="blue")
    plt.axvline(mu_x1_given_x2, color="red", linestyle="--", label="Mean")
    plt.axvline(mu_x1_given_x2 + 2 * math.sqrt(Sigma_x1_given_x2), color="red", linestyle="--", alpha=0.5, label="Mean ± 2σ")
    plt.axvline(mu_x1_given_x2 - 2 * math.sqrt(Sigma_x1_given_x2), color="red", linestyle="--", alpha=0.5)
    plt.title("Figure 6.9c: Conditional Distribution p(x1|x2=-1)")
    plt.xlabel("x1")
    plt.ylabel("p(x1|x2=-1)")
    plt.legend()
    plt.show()

def plot_mixture_of_gaussians():
    """Plot the mixture of Gaussian densities along with components."""
    alpha = 0.4
    mu1, sigma1_2 = 0.0, 0.5
    mu2, sigma2_2 = 2.0, 1.0

    # Generate points for the PDFs
    x = np.linspace(-3, 4, 100)
    p1 = [univariate_gaussian_pdf(xi, mu1, sigma1_2) for xi in x]
    p2 = [univariate_gaussian_pdf(xi, mu2, sigma2_2) for xi in x]
    mixture = [mixture_gaussian_pdf(xi, alpha, mu1, sigma1_2, mu2, sigma2_2) for xi in x]

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=p1, label=f"p1(x) N({mu1}, {sigma1_2})", color="teal")
    sns.lineplot(x=x, y=p2, label=f"p2(x) N({mu2}, {sigma2_2})", color="red")
    sns.lineplot(x=x, y=mixture, label="Mixture p(x)", color="blue")
    plt.title("Mixture of Gaussian Densities")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Gaussian Distribution Visualization with Seaborn")
    print("=" * 60)

    # Plot all figures
    print("Plotting Univariate Gaussian (Figure 6.8a)")
    plot_univariate_gaussian()

    print("\nPlotting Marginal and Conditional Distributions (Figures 6.9b, 6.9c)")
    plot_marginal_and_conditional()

    print("\nPlotting Mixture of Gaussian Densities (Sections 6.5.2, 6.5.3)")
    plot_mixture_of_gaussians()

    print("\n" + "=" * 60)
    print("Summary of Visualizations:")
    print("• Plotted Univariate Gaussian with samples (Figure 6.8a)")
    print("• Plotted Marginal Distribution p(x1) (Figure 6.9b)")
    print("• Plotted Conditional Distribution p(x1|x2=-1) (Figure 6.9c)")
    print("• Plotted Mixture of Gaussian Densities")
```

For a mixture density, the individual components can be considered to be conditional distributions (conditioned on the component identity). Equation (6.85c) is an example of the *conditional variance formula*, also known as the *law of total variance*, which generally states that for two random variables $X$ and $Y$ it holds that

$$
\text{V}_X[x] = \mathbb{E}_Y[\text{V}_X[x|y]] + \text{V}_Y[\mathbb{E}_X[x|y]],
$$

i.e., the (total) variance of $X$ is the expected conditional variance plus the variance of a conditional mean.

We consider in Example 6.17 a bivariate standard Gaussian random variable $X$ and performed a linear transformation $Ax$ on it. The outcome is a Gaussian random variable with mean zero and covariance $AA^\top$. Observe that adding a constant vector will change the mean of the distribution, without affecting its variance, that is, the random variable $x + \mu$ is Gaussian with mean $\mu$ and identity covariance. Hence, any linear/affine transformation of a Gaussian random variable is Gaussian distributed.

Consider a Gaussian distributed random variable $X \sim \mathcal{N}(\mu, \Sigma)$. For a given matrix $A$ of appropriate shape, let $Y$ be a random variable such that $y = Ax$ is a transformed version of $x$. We can compute the mean of $y$ by exploiting that the expectation is a linear operator (6.50) as follows:

$$
\mathbb{E}[y] = \mathbb{E}[Ax] = A \mathbb{E}[x] = A \mu. \tag{6.86}
$$

Similarly, the variance of $y$ can be found by using (6.51):

$$
\text{V}[y] = \text{V}[Ax] = A \text{V}[x] A^\top = A \Sigma A^\top. \tag{6.87}
$$

This means that the random variable $y$ is distributed according to

$$
p(y) = \mathcal{N}(y \mid A \mu, A \Sigma A^\top). \tag{6.88}
$$

Let us now consider the reverse transformation: when we know that a random variable has a mean that is a linear transformation of another random variable. For a given full rank matrix $A \in \mathbb{R}^{M \times N}$, where $M \geqslant N$, let $y \in \mathbb{R}^M$ be a Gaussian random variable with mean $Ax$, i.e.,

$$
p(y) = \mathcal{N}(y \mid Ax, \Sigma). \tag{6.89}
$$

What is the corresponding probability distribution $p(x)$? If $A$ is invertible, then we can write $x = A^{-1} y$ and apply the transformation in the previous paragraph. However, in general $A$ is not invertible, and we use an approach similar to that of the pseudo-inverse (3.57). That is, we premultiply both sides with $A^\top$ and then invert $A^\top A$, which is symmetric and positive definite, giving us the relation

$$
y = Ax \quad \Longleftrightarrow \quad (A^\top A)^{-1} A^\top y = x. \tag{6.90}
$$

Hence, $x$ is a linear transformation of $y$, and we obtain

$$
p(x) = \mathcal{N}\left(x \mid (A^\top A)^{-1} A^\top y, (A^\top A)^{-1} A^\top \Sigma A (A^\top A)^{-1}\right). \tag{6.91}
$$

##  Sampling from Multivariate Gaussian Distributions

We will not explain the subtleties of random sampling on a computer, and the interested reader is referred to Gentle (2004). In the case of a multivariate Gaussian, this process consists of three stages: first, we need a source of pseudo-random numbers that provide a uniform sample in the interval $[0,1]$; second, we use a non-linear transformation such as the Box-Müller transform (Devroye, 1986) to obtain a sample from a univariate Gaussian; and third, we collate a vector of these samples to obtain a sample from a multivariate standard normal $\mathcal{N}(0, I)$.

For a general multivariate Gaussian, that is, where the mean is non-zero and the covariance is not the identity matrix, we use the properties of linear transformations of a Gaussian random variable. Assume we are interested in generating samples $x_i$, $i = 1, \ldots, n$, from a multivariate Gaussian distribution with mean $\mu$ and covariance matrix $\Sigma$. We would like to construct the sample from a sampler that provides samples from the multivariate standard normal $\mathcal{N}(0, I)$.

To obtain samples from a multivariate normal $\mathcal{N}(\mu, \Sigma)$, we can use the properties of a linear transformation of a Gaussian random variable: If $x \sim \mathcal{N}(0, I)$, then $y = Ax + \mu$, where $AA^\top = \Sigma$ is Gaussian distributed with mean $\mu$ and covariance matrix $\Sigma$. One convenient choice of $A$ is to use the Cholesky decomposition (Section 4.3) of the covariance matrix $\Sigma = AA^\top$. The Cholesky decomposition has the benefit that $A$ is triangular, leading to efficient computation.

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # Needed for Seaborn plotting and matrix operations

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Gaussian Distribution Functions ---
def univariate_gaussian_pdf(x, mu, sigma2):
    """Compute the univariate Gaussian PDF (Equation 6.62)."""
    coeff = 1 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

# --- Mixture of Gaussian Densities ---
def mixture_gaussian_pdf(x, alpha, mu1, sigma1_2, mu2, sigma2_2):
    """Compute the PDF of a mixture of two Gaussian densities (Equation 6.80)."""
    p1 = univariate_gaussian_pdf(x, mu1, sigma1_2)
    p2 = univariate_gaussian_pdf(x, mu2, sigma2_2)
    return alpha * p1 + (1 - alpha) * p2

# --- Cholesky Decomposition (Simplified for 2x2 Matrix) ---
def cholesky_decomposition(Sigma):
    """Compute the Cholesky decomposition of a 2x2 covariance matrix (Sigma = LL^T)."""
    Sigma = np.array(Sigma)
    L = np.zeros_like(Sigma)
    L[0, 0] = math.sqrt(Sigma[0, 0])
    L[1, 0] = Sigma[1, 0] / L[0, 0]
    L[1, 1] = math.sqrt(Sigma[1, 1] - L[1, 0] ** 2)
    return L

# --- Generate Samples from Multivariate Standard Normal ---
def generate_standard_normal_samples(n_samples, dim=2):
    """Generate samples from a multivariate standard normal N(0, I) using Box-Muller transform."""
    samples = [[random.gauss(0, 1) for _ in range(dim)] for _ in range(n_samples)]
    return np.array(samples)

# --- Linear Transformation and Sampling ---
def transform_and_sample(mu, Sigma, A, n_samples=1000):
    """
    Generate samples from N(mu, Sigma) by transforming samples from N(0, I).
    y = Ax + mu, where AA^T = Sigma (Section 6.5.4).
    Also compute the transformed distribution (Equations 6.86-6.88).
    """
    # Cholesky decomposition: Sigma = LL^T
    L = cholesky_decomposition(Sigma)

    # Generate samples from N(0, I)
    Z = generate_standard_normal_samples(n_samples, dim=len(mu))

    # Transform: Y = LZ + mu
    Y = Z @ L.T + np.array(mu)

    # Compute theoretical transformed distribution (Equations 6.86-6.88 with A)
    A = np.array(A)
    mu_y = A @ np.array(mu)  # Equation 6.86
    Sigma_y = A @ np.array(Sigma) @ A.T  # Equation 6.87

    return Y, mu_y, Sigma_y

# --- Reverse Transformation (Equations 6.89-6.91) ---
def reverse_transformation(y_samples, A, Sigma):
    """
    Compute the distribution of x given y ~ N(Ax, Sigma) (Equations 6.89-6.91).
    Returns samples of x and the theoretical distribution of x.
    """
    A = np.array(A)
    Sigma = np.array(Sigma)
    y_samples = np.array(y_samples)

    # Compute (A^T A)^(-1) A^T
    A_T = A.T
    A_T_A = A_T @ A
    A_T_A_inv = np.linalg.inv(A_T_A)
    transform_matrix = A_T_A_inv @ A_T

    # Transform samples: x = (A^T A)^(-1) A^T y (Equation 6.90)
    x_samples = y_samples @ transform_matrix.T

    # Compute theoretical distribution of x (Equation 6.91)
    mu_x = transform_matrix @ y_samples.mean(axis=0)
    Sigma_x = transform_matrix @ Sigma @ transform_matrix.T

    return x_samples, mu_x, Sigma_x

# --- Demonstration ---
def demonstrate_gaussian_properties():
    """
    Demonstrate properties of Gaussian distributions (Sections 6.5.3, 6.5.4).
    - Compute and plot mixture of Gaussians (Equations 6.81-6.85)
    - Apply linear transformation (Equations 6.86-6.88)
    - Perform reverse transformation (Equations 6.89-6.91)
    - Sample from multivariate Gaussian (Section 6.5.4)
    """
    print("=== Gaussian Distribution Properties ===")
    print("Sections 6.5.3, 6.5.4: Mixtures, Transformations, and Sampling\n")

    # Step 1: Mixture of Gaussian Densities (Equations 6.81-6.85)
    print("Step 1: Mixture of Gaussian Densities (Equations 6.81-6.85)")
    alpha = 0.4
    mu1, sigma1_2 = 0.0, 0.5
    mu2, sigma2_2 = 2.0, 1.0
    # Compute theoretical mean and variance
    mean_mixture = alpha * mu1 + (1 - alpha) * mu2  # Equation 6.81
    term1 = alpha * sigma1_2 + (1 - alpha) * sigma2_2
    term2 = alpha * (mu1 ** 2) + (1 - alpha) * (mu2 ** 2)
    term3 = (alpha * mu1 + (1 - alpha) * mu2) ** 2
    var_mixture = term1 + term2 - term3  # Equation 6.85c
    print(f"Mixture: p(x) = {alpha} N(x|{mu1}, {sigma1_2}) + {1-alpha} N(x|{mu2}, {sigma2_2})")
    print(f"Theoretical mean of mixture: {mean_mixture:.4f}")
    print(f"Theoretical variance of mixture: {var_mixture:.4f}")
    print()

    # Step 2: Linear Transformation (Equations 6.86-6.88)
    print("Step 2: Linear Transformation (Equations 6.86-6.88)")
    mu = [1, 2]  # Mean of the original Gaussian
    Sigma = [[1, 0.5], [0.5, 1]]  # Covariance matrix
    A = [[2, 1], [1, 3]]  # Transformation matrix
    Y, mu_y, Sigma_y = transform_and_sample(mu, Sigma, A)
    print(f"Original distribution: N({mu}, {Sigma})")
    print(f"Transformation matrix A: {A}")
    print(f"Transformed distribution: N({mu_y.tolist()}, {Sigma_y.tolist()})")
    print()

    # Step 3: Reverse Transformation (Equations 6.89-6.91)
    print("Step 3: Reverse Transformation (Equations 6.89-6.91)")
    # Assume y_samples are from N(Ax, Sigma_y)
    x_samples, mu_x, Sigma_x = reverse_transformation(Y, A, Sigma_y)
    print(f"Reconstructed distribution of x: N({mu_x.tolist()}, {Sigma_x.tolist()})")
    print()

    # Step 4: Sampling from Multivariate Gaussian (Section 6.5.4)
    print("Step 4: Sampling from Multivariate Gaussian (Section 6.5.4)")
    print(f"Generated {len(Y)} samples from N({mu}, {Sigma})")
    print()

# --- Plotting Functions ---
def plot_mixture_of_gaussians():
    """Plot the mixture of Gaussian densities along with components."""
    alpha = 0.4
    mu1, sigma1_2 = 0.0, 0.5
    mu2, sigma2_2 = 2.0, 1.0

    # Generate points for the PDFs
    x = np.linspace(-3, 4, 100)
    p1 = [univariate_gaussian_pdf(xi, mu1, sigma1_2) for xi in x]
    p2 = [univariate_gaussian_pdf(xi, mu2, sigma2_2) for xi in x]
    mixture = [mixture_gaussian_pdf(xi, alpha, mu1, sigma1_2, mu2, sigma2_2) for xi in x]

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=p1, label=f"p1(x) N({mu1}, {sigma1_2})", color="teal")
    sns.lineplot(x=x, y=p2, label=f"p2(x) N({mu2}, {sigma2_2})", color="red")
    sns.lineplot(x=x, y=mixture, label="Mixture p(x)", color="blue")
    plt.title("Mixture of Gaussian Densities")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

def plot_transformed_distribution(Y, mu_y, Sigma_y):
    """Plot the transformed distribution (bivariate Gaussian)."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=Y[:, 0], y=Y[:, 1], alpha=0.5, color="purple", label="Transformed Samples")
    plt.axvline(mu_y[0], color="red", linestyle="--", label="Mean")
    plt.axhline(mu_y[1], color="red", linestyle="--")
    plt.title("Transformed Bivariate Gaussian Distribution")
    plt.xlabel("y1")
    plt.ylabel("y2")
    plt.legend()
    plt.show()

def plot_reconstructed_distribution(x_samples, mu_x, Sigma_x):
    """Plot the reconstructed distribution of x."""
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_samples[:, 0], y=x_samples[:, 1], alpha=0.5, color="green", label="Reconstructed Samples")
    plt.axvline(mu_x[0], color="red", linestyle="--", label="Mean")
    plt.axhline(mu_x[1], color="red", linestyle="--")
    plt.title("Reconstructed Bivariate Gaussian Distribution")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Gaussian Distribution Analysis with Seaborn Plotting")
    print("=" * 60)

    # Run demonstration
    demonstrate_gaussian_properties()

    # Plot figures
    print("Plotting Mixture of Gaussian Densities")
    plot_mixture_of_gaussians()

    print("\nPlotting Transformed Distribution")
    mu = [1, 2]
    Sigma = [[1, 0.5], [0.5, 1]]
    A = [[2, 1], [1, 3]]
    Y, mu_y, Sigma_y = transform_and_sample(mu, Sigma, A)
    plot_transformed_distribution(Y, mu_y, Sigma_y)

    print("\nPlotting Reconstructed Distribution")
    x_samples, mu_x, Sigma_x = reverse_transformation(Y, A, Sigma_y)
    plot_reconstructed_distribution(x_samples, mu_x, Sigma_x)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print("• Computed mean and variance of a mixture of Gaussians (Equations 6.81-6.85)")
    print("• Applied linear transformation to a Gaussian (Equations 6.86-6.88)")
    print("• Performed reverse transformation (Equations 6.89-6.91)")
    print("• Generated samples from a multivariate Gaussian (Section 6.5.4)")
    print("• Plotted Mixture of Gaussian Densities")
    print("• Plotted Transformed Bivariate Gaussian Distribution")
    print("• Plotted Reconstructed Bivariate Gaussian Distribution")
```

## Conjugacy and the Exponential Family

Many of the probability distributions “with names” that we find in statistics textbooks were discovered to model particular types of phenomena. For example, we have seen the Gaussian distribution in Section 6.5. The distributions are also related to each other in complex ways (Leemis and McQueston, 2008). For a beginner in the field, it can be overwhelming to figure out which distribution to use. In addition, many of these distributions were discovered at a time that statistics and computation were done by pencil and paper. *“Computers” used to be a job description.* It is natural to ask what are meaningful concepts in the computing age (Efron and Hastie, 2016). In the previous section, we saw that many of the operations required for inference can be conveniently calculated when the distribution is Gaussian. It is worth recalling at this point the desiderata for manipulating probability distributions in the machine learning context:

1. There is some “closure property” when applying the rules of probability, e.g., Bayes’ theorem. By closure, we mean that applying a particular operation returns an object of the same type.
2. As we collect more data, we do not need more parameters to describe the distribution.
3. Since we are interested in learning from data, we want parameter estimation to behave nicely.

It turns out that the class of distributions called the *exponential family* provides the right balance of generality while retaining favorable computation and inference properties. Before we introduce the exponential family, let us see three more members of “named” probability distributions, the Bernoulli (Example 6.8), Binomial (Example 6.9), and Beta (Example 6.10) distributions.

### Example (Bernoulli Distribution)

The Bernoulli distribution is a distribution for a single binary random variable $X$ with state $x \in \{0, 1\}$. It is governed by a single continuous parameter $\mu \in [0, 1]$ that represents the probability of $X = 1$. The Bernoulli distribution $\text{Ber}(\mu)$ is defined as

$$
p(x \mid \mu) = \mu^x (1 - \mu)^{1-x}, \quad x \in \{0, 1\}, \tag{6.92}
$$

$$
\mathbb{E}[x] = \mu, \tag{6.93}
$$

$$
\text{V}[x] = \mu (1 - \mu), \tag{6.94}
$$

where $\mathbb{E}[x]$ and $\text{V}[x]$ are the mean and variance of the binary random variable $X$. An example where the Bernoulli distribution can be used is when we are interested in modeling the probability of “heads” when flipping a coin.

The x-axis represents the number $m$ of observations $x = 1$ in $N = 15$ experiments, and the y-axis represents the probability $p(m)$.

### Example (Binomial Distribution)

The Binomial distribution is a generalization of the Bernoulli distribution to a distribution over integers (illustrated in Figure 6.10). In particular, the Binomial can be used to describe the probability of observing $m$ occurrences of $X = 1$ in a set of $N$ samples from a Bernoulli distribution where $p(X = 1) = \mu \in [0, 1]$. The Binomial distribution $\text{Bin}(N, \mu)$ is defined as

$$
p(m \mid N, \mu) = \binom{N}{m} \mu^m (1 - \mu)^{N - m}, \tag{6.95}
$$

$$
\mathbb{E}[m] = N \mu, \tag{6.96}
$$

$$
\text{V}[m] = N \mu (1 - \mu), \tag{6.97}
$$

where $\mathbb{E}[m]$ and $\text{V}[m]$ are the mean and variance of $m$, respectively. An example where the Binomial could be used is if we want to describe the probability of observing $m$ “heads” in $N$ coin-flip experiments if the probability for observing head in a single experiment is $\mu$.

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta  # For Beta distribution PDF and sampling

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Custom Binomial Coefficient Function ---
def binomial_coefficient(n, k):
    """Compute the binomial coefficient n choose k."""
    if k < 0 or k > n:
        return 0
    # Optimize by using the smaller value of k or (n-k)
    k = min(k, n - k)
    numerator = 1
    denominator = 1
    for i in range(k):
        numerator *= (n - i)
        denominator *= (i + 1)
    return numerator // denominator

# --- Bernoulli Distribution (Example 6.8) ---
def bernoulli_pmf(x, mu):
    """Compute the Bernoulli PMF (Equation 6.92)."""
    if x not in [0, 1]:
        return 0
    return mu ** x * (1 - mu) ** (1 - x)

def generate_bernoulli_samples(mu, n_samples=1000):
    """Generate samples from a Bernoulli distribution."""
    return [1 if random.random() < mu else 0 for _ in range(n_samples)]

def compute_mean(samples):
    return sum(samples) / len(samples)

def compute_variance(samples):
    mean = compute_mean(samples)
    return sum((x - mean) ** 2 for x in samples) / len(samples)

# --- Binomial Distribution (Example 6.9) ---
def binomial_pmf(m, N, mu):
    """Compute the Binomial PMF (Equation 6.95)."""
    if m < 0 or m > N:
        return 0
    binomial_coeff = binomial_coefficient(N, m)  # Use custom function
    return binomial_coeff * (mu ** m) * (1 - mu) ** (N - m)

def generate_binomial_samples(N, mu, n_samples=1000):
    """Generate Binomial samples by summing Bernoulli samples."""
    samples = []
    for _ in range(n_samples):
        bernoulli_samples = generate_bernoulli_samples(mu, N)
        samples.append(sum(bernoulli_samples))
    return samples

# --- Demonstration ---
def demonstrate_distributions():
    """
    Demonstrate Bernoulli, Binomial, and Beta distributions (Section 6.6).
    - Bernoulli: Generate samples, compute mean and variance (Equations 6.92-6.94)
    - Binomial: Generate samples, compute mean and variance, plot for different mu (Equations 6.95-6.97, Figure 6.10)
    - Beta: Generate samples and plot the density (Example 6.10)
    """
    print("=== Probability Distributions Analysis ===")
    print("Section 6.6: Bernoulli, Binomial, and Beta Distributions\n")

    # Step 1: Bernoulli Distribution (Example 6.8)
    print("Step 1: Bernoulli Distribution (Example 6.8)")
    mu_bernoulli = 0.4
    bernoulli_samples = generate_bernoulli_samples(mu_bernoulli)
    empirical_mean = compute_mean(bernoulli_samples)
    empirical_var = compute_variance(bernoulli_samples)
    theoretical_mean = mu_bernoulli  # Equation 6.93
    theoretical_var = mu_bernoulli * (1 - mu_bernoulli)  # Equation 6.94
    print(f"Bernoulli with mu = {mu_bernoulli}")
    print(f"Theoretical mean: {theoretical_mean:.4f}")
    print(f"Empirical mean: {empirical_mean:.4f}")
    print(f"Theoretical variance: {theoretical_var:.4f}")
    print(f"Empirical variance: {empirical_var:.4f}")
    print()

    # Step 2: Binomial Distribution (Example 6.9)
    print("Step 2: Binomial Distribution (Example 6.9)")
    N = 15
    mu_values = [0.1, 0.4, 0.75]
    for mu in mu_values:
        binomial_samples = generate_binomial_samples(N, mu)
        empirical_mean = compute_mean(binomial_samples)
        empirical_var = compute_variance(binomial_samples)
        theoretical_mean = N * mu  # Equation 6.96
        theoretical_var = N * mu * (1 - mu)  # Equation 6.97
        print(f"Binomial with N = {N}, mu = {mu}")
        print(f"Theoretical mean: {theoretical_mean:.4f}")
        print(f"Empirical mean: {empirical_mean:.4f}")
        print(f"Theoretical variance: {theoretical_var:.4f}")
        print(f"Empirical variance: {empirical_var:.4f}")
        print()

    # Step 3: Beta Distribution (Example 6.10)
    print("Step 3: Beta Distribution (Example 6.10)")
    alpha, beta_param = 2, 5  # Parameters for the Beta distribution
    print(f"Beta distribution with alpha = {alpha}, beta = {beta_param}")
    print()

# --- Plotting Functions ---
def plot_bernoulli_distribution():
    """Plot the Bernoulli distribution probabilities."""
    mu = 0.4
    x = [0, 1]
    probs = [bernoulli_pmf(xi, mu) for xi in x]

    plt.figure(figsize=(8, 6))
    sns.barplot(x=x, y=probs, color="teal")
    plt.title(f"Bernoulli Distribution (mu = {mu})")
    plt.xlabel("x")
    plt.ylabel("p(x)")
    plt.xticks([0, 1], ["0", "1"])
    plt.show()

def plot_binomial_distribution():
    """Plot the Binomial distribution for different mu values (Figure 6.10)."""
    N = 15
    mu_values = [0.1, 0.4, 0.75]
    m = np.arange(0, N + 1)

    plt.figure(figsize=(10, 6))
    for mu in mu_values:
        probs = [binomial_pmf(mi, N, mu) for mi in m]
        sns.barplot(x=m, y=probs, label=f"mu = {mu}", alpha=0.6)

    plt.title("Figure 6.10: Binomial Distribution (N = 15)")
    plt.xlabel("Number m of observations x = 1")
    plt.ylabel("p(m)")
    plt.legend()
    plt.show()

def plot_beta_distribution():
    """Plot the Beta distribution density."""
    alpha, beta_param = 2, 5
    x = np.linspace(0, 1, 100)
    pdf = beta.pdf(x, alpha, beta_param)

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x, y=pdf, color="purple", label=f"Beta({alpha}, {beta_param})")
    plt.title("Beta Distribution")
    plt.xlabel("mu")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Probability Distributions Analysis with Seaborn Plotting")
    print("=" * 60)

    # Run demonstration
    demonstrate_distributions()

    # Plot figures
    print("Plotting Bernoulli Distribution")
    plot_bernoulli_distribution()

    print("\nPlotting Binomial Distribution (Figure 6.10)")
    plot_binomial_distribution()

    print("\nPlotting Beta Distribution")
    plot_beta_distribution()

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print("• Demonstrated Bernoulli distribution properties (Example 6.8)")
    print("• Demonstrated Binomial distribution properties for different mu (Example 6.9, Figure 6.10)")
    print("• Demonstrated Beta distribution (Example 6.10)")
    print("• Plotted Bernoulli Distribution")
    print("• Plotted Binomial Distribution (Figure 6.10)")
    print("• Plotted Beta Distribution")
```

distribution $\text{Beta}(\alpha, \beta)$ (illustrated in Figure 6.11) itself is governed by two parameters $\alpha > 0$, $\beta > 0$ and is defined as

$$
p(\mu \mid \alpha, \beta) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} \mu^{\alpha - 1} (1 - \mu)^{\beta - 1}, \tag{6.98}
$$

$$
\mathbb{E}[\mu] = \frac{\alpha}{\alpha + \beta}, \quad \text{V}[\mu] = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}, \tag{6.99}
$$

where $\Gamma(\cdot)$ is the Gamma function defined as

$$
\Gamma(t) := \int_{0}^{\infty} x^{t-1} \exp(-x) \, dx, \quad t > 0, \tag{6.100}
$$

$$
\Gamma(t + 1) = t \Gamma(t). \tag{6.101}
$$

Note that the fraction of Gamma functions in (6.98) normalizes the Beta distribution.

- $\alpha = 2$, $\beta = 0.3$

The x-axis represents $\mu \in [0, 1]$, and the y-axis represents the probability density $p(\mu \mid \alpha, \beta)$.

Intuitively, $\alpha$ moves probability mass toward 1, whereas $\beta$ moves probability mass toward 0. There are some special cases (Murphy, 2012):

- For $\alpha = 1 = \beta$, we obtain the uniform distribution $U[0, 1]$.
- For $\alpha, \beta < 1$, we get a bimodal distribution with spikes at 0 and 1.
- For $\alpha, \beta > 1$, the distribution is unimodal.
- For $\alpha, \beta > 1$ and $\alpha = \beta$, the distribution is unimodal, symmetric, and centered in the interval $[0, 1]$, i.e., the mode/mean is at $\frac{1}{2}$.

## 6.6.1 Conjugacy

According to Bayes’ theorem (6.23), the posterior is proportional to the product of the prior and the likelihood. The specification of the prior can be tricky for two reasons: First, the prior should encapsulate our knowledge about the problem before we see any data. This is often difficult to describe. Second, it is often not possible to compute the posterior distribution analytically. However, there are some priors that are computationally convenient: *conjugate priors*.

### Definition 6.13 (Conjugate Prior)

A prior is *conjugate* for the likelihood function if the posterior is of the same form/type as the prior.

Conjugacy is particularly convenient because we can algebraically calculate our posterior distribution by updating the parameters of the prior distribution.

### Remark

When considering the geometry of probability distributions, conjugate priors retain the same distance structure as the likelihood (Agarwal and Daumé III, 2010). $\diamond$

To introduce a concrete example of conjugate priors, we describe in Example 6.11 the Binomial distribution (defined on discrete random variables) and the Beta distribution (defined on continuous random variables).

### Example 6.11 (Beta-Binomial Conjugacy)

Consider a Binomial random variable $x \sim \text{Bin}(N, \mu)$ where

$$
p(x \mid N, \mu) = \binom{N}{x} \mu^x (1 - \mu)^{N - x}, \quad x = 0, 1, \ldots, N, \tag{6.102}
$$

is the probability of finding $x$ times the outcome “heads” in $N$ coin flips, where $\mu$ is the probability of a “head”. We place a Beta prior on the parameter $\mu$, that is, $\mu \sim \text{Beta}(\alpha, \beta)$, where

$$
p(\mu \mid \alpha, \beta) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} \mu^{\alpha - 1} (1 - \mu)^{\beta - 1}. \tag{6.103}
$$

If we now observe some outcome $x = h$, that is, we see $h$ heads in $N$ coin flips, we compute the posterior distribution on $\mu$ as

$$
p(\mu \mid x = h, N, \alpha, \beta) \propto p(x \mid N, \mu) p(\mu \mid \alpha, \beta) \tag{6.104a}
$$

$$
\propto \mu^h (1 - \mu)^{(N - h)} \mu^{\alpha - 1} (1 - \mu)^{\beta - 1} \tag{6.104b}
$$

$$
= \mu^{h + \alpha - 1} (1 - \mu)^{(N - h) + \beta - 1} \tag{6.104c}
$$


$$
\propto \text{Beta}(h + \alpha, N - h + \beta), \tag{6.104d}
$$

i.e., the posterior distribution is a Beta distribution as the prior, i.e., the Beta prior is conjugate for the parameter $\mu$ in the Binomial likelihood function.

In the following example, we will derive a result that is similar to the Beta-Binomial conjugacy result. Here we will show that the Beta distribution is a conjugate prior for the Bernoulli distribution.

### Example 6.12 (Beta-Bernoulli Conjugacy)

Let $x \in \{0, 1\}$ be distributed according to the Bernoulli distribution with parameter $\theta \in [0, 1]$, that is, $p(x = 1 \mid \theta) = \theta$. This can also be expressed as

$$
p(x \mid \theta) = \theta^x (1 - \theta)^{1 - x}.
$$

Let $\theta$ be distributed according to a Beta distribution with parameters $\alpha, \beta$, that is,

$$
p(\theta \mid \alpha, \beta) \propto \theta^{\alpha - 1} (1 - \theta)^{\beta - 1}.
$$

Multiplying the Beta and the Bernoulli distributions, we get

$$
p(\theta \mid x, \alpha, \beta) \propto p(x \mid \theta) p(\theta \mid \alpha, \beta) \tag{6.105a}
$$

$$
= \theta^x (1 - \theta)^{1 - x} \theta^{\alpha - 1} (1 - \theta)^{\beta - 1} \tag{6.105b}
$$

$$
= \theta^{\alpha + x - 1} (1 - \theta)^{\beta + (1 - x) - 1} \tag{6.105c}
$$

$$
\propto p(\theta \mid \alpha + x, \beta + (1 - x)). \tag{6.105d}
$$

The last line is the Beta distribution with parameters $(\alpha + x, \beta + (1 - x))$.

### Table 6.2: Examples of Conjugate Priors for Common Likelihood Functions

| Likelihood       | Conjugate Prior         | Posterior             |
|------------------|-------------------------|-----------------------|
| Bernoulli        | Beta                    | Beta                  |
| Binomial         | Beta                    | Beta                  |
| Gaussian         | Gaussian/inverse Gamma  | Gaussian/inverse Gamma|
| Gaussian         | Gaussian/inverse Wishart| Gaussian/inverse Wishart|
| Multinomial      | Dirichlet               | Dirichlet             |

*The Gamma prior is conjugate for the precision (inverse variance) in the univariate Gaussian likelihood, and the Wishart prior is conjugate for the precision matrix (inverse covariance matrix) in the multivariate Gaussian likelihood.*

The Beta distribution is the conjugate prior for the parameter $\mu$ in both the Binomial and the Bernoulli likelihood. For a Gaussian likelihood function, we can place a conjugate Gaussian prior on the mean. The reason why the Gaussian likelihood appears twice in the table is that we need to distinguish the univariate from the multivariate case. In the univariate (scalar) case, the inverse Gamma is the conjugate prior for the variance. In the multivariate case, we use a conjugate inverse Wishart distribution as a prior on the covariance matrix. The Dirichlet distribution is the conjugate prior for the Multinomial likelihood.

## 6.6.2 Sufficient Statistics

Recall that a statistic of a random variable is a deterministic function of that random variable. For example, if $x = [x_1, \ldots, x_N]^\top$ is a vector of univariate Gaussian random variables, that is, $x_n \sim \mathcal{N}(\mu, \sigma^2)$, then the sample mean $\hat{\mu} = \frac{1}{N} (x_1 + \cdots + x_N)$ is a statistic. Sir Ronald Fisher discovered the notion of *sufficient statistics*: the idea that there are statistics that will contain all available information that can be inferred from data corresponding to the distribution under consideration. In other words, sufficient statistics carry all the information needed to make inference about the population, that is, they are the statistics that are *sufficient* to represent the distribution.

For a set of distributions parametrized by $\theta$, let $X$ be a random variable with distribution $p(x \mid \theta_0)$ given an unknown $\theta_0$. A vector $\phi(x)$ of statistics is called *sufficient statistics* for $\theta_0$ if they contain all possible information about $\theta_0$. To be more formal about “contain all possible information”, this means that the probability of $x$ given $\theta$ can be factored into a part that does not depend on $\theta$, and a part that depends on $\theta$ only via $\phi(x)$. The Fisher-Neyman factorization theorem formalizes this notion, which we state in Theorem 6.14 without proof.

### Theorem 6.14 (Fisher-Neyman)

[Theorem 6.5 in Lehmann and Casella (1998)] Let $X$ have probability density function $p(x \mid \theta)$. Then the statistics $\phi(x)$ are sufficient for $\theta$ if and only if $p(x \mid \theta)$ can be written in the form

$$
p(x \mid \theta) = h(x) g_\theta(\phi(x)), \tag{6.106}
$$

where $h(x)$ is a distribution independent of $\theta$ and $g_\theta$ captures all the dependence on $\theta$ via sufficient statistics $\phi(x)$.

If $p(x \mid \theta)$ does not depend on $\theta$, then $\phi(x)$ is trivially a sufficient statistic for any function $\phi$. The more interesting case is that $p(x \mid \theta)$ is dependent only on $\phi(x)$ and not $x$ itself. In this case, $\phi(x)$ is a sufficient statistic for $\theta$.

In machine learning, we consider a finite number of samples from a distribution. One could imagine that for simple distributions (such as the Bernoulli in Example 6.8) we only need a small number of samples to estimate the parameters of the distributions. We could also consider the opposite problem: If we have a set of data (a sample from an unknown distribution), which distribution gives the best fit? A natural question to ask is, as we observe more data, do we need more parameters $\theta$ to describe the distribution? It turns out that the answer is yes in general, and this is studied in non-parametric statistics (Wasserman, 2007). A converse question is to consider which class of distributions have finite-dimensional sufficient statistics, that is, the number of parameters needed to describe them does not increase arbitrarily. The answer is exponential family distributions, described in the following section.

import random
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Gamma Function Approximation (for Beta Distribution) ---
def gamma_approx(t):
    """Approximate the Gamma function using Stirling's approximation for t > 0."""
    if t <= 0:
        raise ValueError("Gamma function defined for t > 0")
    # Stirling's approximation: Γ(t) ≈ sqrt(2π/t) * (t/e)^t
    return math.sqrt(2 * math.pi / t) * (t / math.e) ** t

# --- Beta Distribution PDF ---
def beta_pdf(mu, alpha, beta_param):
    """Compute the Beta PDF (Equation 6.98) for mu in [0, 1]."""
    if not (0 <= mu <= 1):
        return 0
    # Normalization constant: Γ(α + β) / (Γ(α)Γ(β))
    norm = gamma_approx(alpha + beta_param) / (gamma_approx(alpha) * gamma_approx(beta_param))
    return norm * (mu ** (alpha - 1)) * ((1 - mu) ** (beta_param - 1))

# --- Bernoulli Distribution ---
def bernoulli_pmf(x, theta):
    """Compute the Bernoulli PMF."""
    if x not in [0, 1]:
        return 0
    return theta ** x * (1 - theta) ** (1 - x)

def generate_bernoulli_samples(theta, n_samples=1):
    """Generate Bernoulli samples."""
    return [1 if random.random() < theta else 0 for _ in range(n_samples)]

# --- Binomial Distribution ---
def binomial_coefficient(n, k):
    """Compute the binomial coefficient n choose k."""
    if k < 0 or k > n:
        return 0
    k = min(k, n - k)
    numerator = 1
    denominator = 1
    for i in range(k):
        numerator *= (n - i)
        denominator *= (i + 1)
    return numerator // denominator

def binomial_pmf(x, N, mu):
    """Compute the Binomial PMF (Equation 6.102)."""
    if x < 0 or x > N:
        return 0
    return binomial_coefficient(N, x) * (mu ** x) * ((1 - mu) ** (N - x))

# --- Gaussian Distribution ---
def gaussian_pdf(x, mu, sigma2):
    """Compute the univariate Gaussian PDF."""
    coeff = 1 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

def generate_gaussian_samples(mu, sigma2, n_samples=1000):
    """Generate samples from a univariate Gaussian using random.gauss."""
    return [random.gauss(mu, math.sqrt(sigma2)) for _ in range(n_samples)]

# --- Compute Sufficient Statistics ---
def compute_sample_mean(samples):
    return sum(samples) / len(samples)

def compute_sample_variance(samples):
    mean = compute_sample_mean(samples)
    return sum((x - mean) ** 2 for x in samples) / len(samples)

# --- Demonstration ---
def demonstrate_conjugacy_and_sufficient_statistics():
    """
    Demonstrate conjugacy and sufficient statistics (Sections 6.6.1, 6.6.2).
    - Beta-Binomial conjugacy (Example 6.11, Equations 6.102-6.104)
    - Beta-Bernoulli conjugacy (Example 6.12, Equations 6.105)
    - Sufficient statistics for Gaussian (Section 6.6.2)
    """
    print("=== Conjugacy and Sufficient Statistics Analysis ===")
    print("Sections 6.6.1, 6.6.2: Beta-Binomial, Beta-Bernoulli, and Gaussian\n")

    # Step 1: Beta-Binomial Conjugacy (Example 6.11)
    print("Step 1: Beta-Binomial Conjugacy (Example 6.11)")
    N = 10  # Number of coin flips
    h = 7   # Number of heads observed
    alpha_prior, beta_prior = 2, 2  # Prior Beta parameters
    # Posterior parameters (Equation 6.104d)
    alpha_posterior = h + alpha_prior
    beta_posterior = N - h + beta_prior
    print(f"Binomial likelihood: Bin({N}, mu) with observed h = {h}")
    print(f"Prior: Beta({alpha_prior}, {beta_prior})")
    print(f"Posterior: Beta({alpha_posterior}, {beta_posterior})")
    print()

    # Step 2: Beta-Bernoulli Conjugacy (Example 6.12)
    print("Step 2: Beta-Bernoulli Conjugacy (Example 6.12)")
    theta = 0.6  # True parameter for simulation
    x = generate_bernoulli_samples(theta, n_samples=1)[0]  # Single observation
    alpha_prior, beta_prior = 1, 1  # Prior Beta parameters (Uniform)
    # Posterior parameters (Equation 6.105d)
    alpha_posterior = alpha_prior + x
    beta_posterior = beta_prior + (1 - x)
    print(f"Bernoulli likelihood: Ber(theta) with observed x = {x}")
    print(f"Prior: Beta({alpha_prior}, {beta_prior})")
    print(f"Posterior: Beta({alpha_posterior}, {beta_posterior})")
    print()

    # Step 3: Sufficient Statistics for Gaussian (Section 6.6.2)
    print("Step 3: Sufficient Statistics for Gaussian (Section 6.6.2)")
    mu, sigma2 = 0, 1  # Gaussian parameters
    samples = generate_gaussian_samples(mu, sigma2, n_samples=1000)
    sample_mean = compute_sample_mean(samples)
    sample_var = compute_sample_variance(samples)
    print(f"Generated {len(samples)} samples from N({mu}, {sigma2})")
    print(f"Sample mean (sufficient statistic): {sample_mean:.4f}")
    print(f"Sample variance (sufficient statistic): {sample_var:.4f}")
    print()

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Gamma Function Approximation (for Beta Distribution) ---
def gamma_approx(t):
    """Approximate the Gamma function using Stirling's approximation for t > 0."""
    if t <= 0:
        raise ValueError("Gamma function defined for t > 0")
    # Stirling's approximation: Γ(t) ≈ sqrt(2π/t) * (t/e)^t
    return math.sqrt(2 * math.pi / t) * (t / math.e) ** t

# --- Beta Distribution PDF ---
def beta_pdf(mu, alpha, beta_param):
    """Compute the Beta PDF (Equation 6.98) for mu in [0, 1]."""
    if not (0 <= mu <= 1):
        return 0
    # Normalization constant: Γ(α + β) / (Γ(α)Γ(β))
    norm = gamma_approx(alpha + beta_param) / (gamma_approx(alpha) * gamma_approx(beta_param))
    return norm * (mu ** (alpha - 1)) * ((1 - mu) ** (beta_param - 1))

# --- Bernoulli Distribution ---
def bernoulli_pmf(x, theta):
    """Compute the Bernoulli PMF."""
    if x not in [0, 1]:
        return 0
    return theta ** x * (1 - theta) ** (1 - x)

def generate_bernoulli_samples(theta, n_samples=1):
    """Generate Bernoulli samples."""
    return [1 if random.random() < theta else 0 for _ in range(n_samples)]

# --- Binomial Distribution ---
def binomial_coefficient(n, k):
    """Compute the binomial coefficient n choose k."""
    if k < 0 or k > n:
        return 0
    k = min(k, n - k)
    numerator = 1
    denominator = 1
    for i in range(k):
        numerator *= (n - i)
        denominator *= (i + 1)
    return numerator // denominator

def binomial_pmf(x, N, mu):
    """Compute the Binomial PMF (Equation 6.102)."""
    if x < 0 or x > N:
        return 0
    return binomial_coefficient(N, x) * (mu ** x) * ((1 - mu) ** (N - x))

# --- Gaussian Distribution ---
def gaussian_pdf(x, mu, sigma2):
    """Compute the univariate Gaussian PDF."""
    coeff = 1 / math.sqrt(2 * math.pi * sigma2)
    exponent = -((x - mu) ** 2) / (2 * sigma2)
    return coeff * math.exp(exponent)

def generate_gaussian_samples(mu, sigma2, n_samples=1000):
    """Generate samples from a univariate Gaussian using random.gauss."""
    return [random.gauss(mu, math.sqrt(sigma2)) for _ in range(n_samples)]

# --- Compute Sufficient Statistics ---
def compute_sample_mean(samples):
    return sum(samples) / len(samples)

def compute_sample_variance(samples):
    mean = compute_sample_mean(samples)
    return sum((x - mean) ** 2 for x in samples) / len(samples)

# --- Demonstration ---
def demonstrate_conjugacy_and_sufficient_statistics():
    """
    Demonstrate conjugacy and sufficient statistics (Sections 6.6.1, 6.6.2).
    - Beta-Binomial conjugacy (Example 6.11, Equations 6.102-6.104)
    - Beta-Bernoulli conjugacy (Example 6.12, Equations 6.105)
    - Sufficient statistics for Gaussian (Section 6.6.2)
    """
    print("=== Conjugacy and Sufficient Statistics Analysis ===")
    print("Sections 6.6.1, 6.6.2: Beta-Binomial, Beta-Bernoulli, and Gaussian\n")

    # Step 1: Beta-Binomial Conjugacy (Example 6.11)
    print("Step 1: Beta-Binomial Conjugacy (Example 6.11)")
    N = 10  # Number of coin flips
    h = 7   # Number of heads observed
    alpha_prior, beta_prior = 2, 2  # Prior Beta parameters
    # Posterior parameters (Equation 6.104d)
    alpha_posterior = h + alpha_prior
    beta_posterior = N - h + beta_prior
    print(f"Binomial likelihood: Bin({N}, mu) with observed h = {h}")
    print(f"Prior: Beta({alpha_prior}, {beta_prior})")
    print(f"Posterior: Beta({alpha_posterior}, {beta_posterior})")
    print()

    # Step 2: Beta-Bernoulli Conjugacy (Example 6.12)
    print("Step 2: Beta-Bernoulli Conjugacy (Example 6.12)")
    theta = 0.6  # True parameter for simulation
    x = generate_bernoulli_samples(theta, n_samples=1)[0]  # Single observation
    alpha_prior, beta_prior = 1, 1  # Prior Beta parameters (Uniform)
    # Posterior parameters (Equation 6.105d)
    alpha_posterior = alpha_prior + x
    beta_posterior = beta_prior + (1 - x)
    print(f"Bernoulli likelihood: Ber(theta) with observed x = {x}")
    print(f"Prior: Beta({alpha_prior}, {beta_prior})")
    print(f"Posterior: Beta({alpha_posterior}, {beta_posterior})")
    print()

    # Step 3: Sufficient Statistics for Gaussian (Section 6.6.2)
    print("Step 3: Sufficient Statistics for Gaussian (Section 6.6.2)")
    mu, sigma2 = 0, 1  # Gaussian parameters
    samples = generate_gaussian_samples(mu, sigma2, n_samples=1000)
    sample_mean = compute_sample_mean(samples)
    sample_var = compute_sample_variance(samples)
    print(f"Generated {len(samples)} samples from N({mu}, {sigma2})")
    print(f"Sample mean (sufficient statistic): {sample_mean:.4f}")
    print(f"Sample variance (sufficient statistic): {sample_var:.4f}")
    print()

# --- Plotting Functions ---
def plot_beta_binomial_conjugacy():
    """Plot the prior and posterior Beta distributions for Beta-Binomial conjugacy."""
    N = 10
    h = 7
    alpha_prior, beta_prior = 2, 2
    alpha_posterior = h + alpha_prior
    beta_posterior = N - h + beta_prior

    mu = [i / 100 for i in range(101)]  # mu from 0 to 1
    prior = [beta_pdf(m, alpha_prior, beta_prior) for m in mu]
    posterior = [beta_pdf(m, alpha_posterior, beta_posterior) for m in mu]

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=mu, y=prior, label=f"Prior: Beta({alpha_prior}, {beta_prior})", color="teal")
    sns.lineplot(x=mu, y=posterior, label=f"Posterior: Beta({alpha_posterior}, {beta_posterior})", color="purple")
    plt.title("Beta-Binomial Conjugacy: Prior and Posterior")
    plt.xlabel("mu")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

def plot_beta_bernoulli_conjugacy():
    """Plot the prior and posterior for Beta-Bernoulli conjugacy."""
    theta = 0.6
    x = generate_bernoulli_samples(theta, n_samples=1)[0]
    alpha_prior, beta_prior = 1, 1
    alpha_posterior = alpha_prior + x
    beta_posterior = beta_prior + (1 - x)

    mu = [i / 100 for i in range(101)]
    prior = [beta_pdf(m, alpha_prior, beta_prior) for m in mu]
    posterior = [beta_pdf(m, alpha_posterior, beta_posterior) for m in mu]

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=mu, y=prior, label=f"Prior: Beta({alpha_prior}, {beta_prior})", color="teal")
    sns.lineplot(x=mu, y=posterior, label=f"Posterior: Beta({alpha_posterior}, {beta_posterior})", color="purple")
    plt.title(f"Beta-Bernoulli Conjugacy: Prior and Posterior (x = {x})")
    plt.xlabel("theta")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

def plot_gaussian_sufficient_statistics():
    """Plot Gaussian samples with the sample mean."""
    mu, sigma2 = 0, 1
    samples = generate_gaussian_samples(mu, sigma2, n_samples=1000)
    sample_mean = compute_sample_mean(samples)

    plt.figure(figsize=(8, 6))
    sns.histplot(samples, stat="density", color="orange", alpha=0.5, label="Samples")
    plt.axvline(sample_mean, color="red", linestyle="--", label=f"Sample Mean: {sample_mean:.2f}")
    plt.title("Gaussian Samples with Sufficient Statistic (Sample Mean)")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Conjugacy and Sufficient Statistics Analysis with Seaborn Plotting")
    print("=" * 60)

    # Run demonstration
    demonstrate_conjugacy_and_sufficient_statistics()

    # Plot figures
    print("Plotting Beta-Binomial Conjugacy")
    plot_beta_binomial_conjugacy()

    print("\nPlotting Beta-Bernoulli Conjugacy")
    plot_beta_bernoulli_conjugacy()

    print("\nPlotting Gaussian Sufficient Statistics")
    plot_gaussian_sufficient_statistics()

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print("• Demonstrated Beta-Binomial conjugacy (Example 6.11)")
    print("• Demonstrated Beta-Bernoulli conjugacy (Example 6.12)")
    print("• Computed sufficient statistics for Gaussian (Section 6.6.2)")
    print("• Plotted Beta-Binomial Prior and Posterior")
    print("• Plotted Beta-Bernoulli Prior and Posterior")
    print("• Plotted Gaussian Samples with Sample Mean")
```
