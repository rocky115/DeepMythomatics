### 6.6.3 Exponential Family

There are three possible levels of abstraction we can have when considering distributions (of discrete or continuous random variables). At level one (the most concrete end of the spectrum), we have a particular named distribution with fixed parameters, for example a univariate Gaussian $\mathcal{N}(0, 1)$ with zero mean and unit variance. In machine learning, we often use the second level of abstraction, that is, we fix the parametric form (the univariate Gaussian) and infer the parameters from data. For example, we assume a univariate Gaussian $\mathcal{N}(\mu, \sigma^2)$ with unknown mean $\mu$ and unknown variance $\sigma^2$, and use a maximum likelihood fit to determine the best parameters $(\mu, \sigma^2)$. We will see an example of this when considering linear regression in Chapter 9. A third level of abstraction is to consider families of distributions, and in this book, we consider the **exponential family**. The univariate Gaussian is an example of a member of the exponential family. Many of the widely used statistical models, including all the “named” models in Table 6.2, are members of the exponential family. They can all be unified into one concept (Brown, 1986).

**Remark.** A brief historical anecdote: Like many concepts in mathematics and science, exponential families were independently discovered at the same time by different researchers. In the years 1935–1936, Edwin Pitman in Tasmania, Georges Darmois in Paris, and Bernard Koopman in New York independently showed that the exponential families are the only families that enjoy finite-dimensional sufficient statistics under repeated independent sampling (Lehmann and Casella, 1998). $\diamondsuit$

An **exponential family** is a family of probability distributions, parameterized by $\boldsymbol{\theta} \in \mathbb{R}^D$, of the form:

$$ p(\mathbf{x} | \boldsymbol{\theta}) = h(\mathbf{x}) \exp (\langle\boldsymbol{\theta}, \boldsymbol{\phi}(\mathbf{x})\rangle - A(\boldsymbol{\theta})) \quad \text{(6.107)} $$

where $\boldsymbol{\phi}(\mathbf{x})$ is the vector of sufficient statistics. In general, any inner product (Section 3.2) can be used in (6.107), and for concreteness we will use the standard dot product here ($\langle\boldsymbol{\theta}, \boldsymbol{\phi}(\mathbf{x})\rangle = \boldsymbol{\theta}^\top \boldsymbol{\phi}(\mathbf{x})$). Note that the form of the exponential family is essentially a particular expression of $g_{\boldsymbol{\theta}}(\boldsymbol{\phi}(\mathbf{x}))$ in the Fisher-Neyman theorem (Theorem 6.14). The factor $h(\mathbf{x})$ can be absorbed into the dot product term by adding another entry ($\log h(\mathbf{x})$) to the vector of sufficient statistics $\boldsymbol{\phi}(\mathbf{x})$, and constraining the corresponding parameter $\theta_0 = 1$. The term $A(\boldsymbol{\theta})$ is the normalization constant that ensures that the distribution sums up or integrates to one and is called the **log-partition function**. A good intuitive notion of exponential families can be obtained by ignoring these two terms and considering exponential families as distributions of the form:

$$ p(\mathbf{x} | \boldsymbol{\theta}) \propto \exp(\boldsymbol{\theta}^\top \boldsymbol{\phi}(\mathbf{x})) \quad \text{(6.108)} $$

For this form of parametrization, the parameters $\boldsymbol{\theta}$ are called the **natural parameters**. At first glance, it seems that exponential families are a mundane transformation by adding the exponential function to the result of a dot product. However, there are many implications that allow for convenient modeling and efficient computation based on the fact that we can capture information about data in $\boldsymbol{\phi}(\mathbf{x})$.

**Example 6.13 (Gaussian as Exponential Family)**

Consider the univariate Gaussian distribution $\mathcal{N}(\mu, \sigma^2)$. Let $\boldsymbol{\phi}(x) = \begin{bmatrix} x \\ x^2 \end{bmatrix}$. Then by using the definition of the exponential family,

$$ p(x | \boldsymbol{\theta}) \propto \exp(\theta_1 x + \theta_2 x^2) \quad \text{(6.109)} $$

Setting $\boldsymbol{\theta} = \begin{bmatrix} \mu/\sigma^2 \\ -1/(2\sigma^2) \end{bmatrix}$ and substituting into (6.109), we obtain:

$$ p(x | \boldsymbol{\theta}) \propto \exp\left( \frac{\mu x}{\sigma^2} - \frac{x^2}{2\sigma^2} \right) \propto \exp\left( -\frac{1}{2\sigma^2} (x - \mu)^2 \right) \quad \text{(6.111)} $$

Therefore, the univariate Gaussian distribution is a member of the exponential family with sufficient statistic $\boldsymbol{\phi}(x) = \begin{bmatrix} x \\ x^2 \end{bmatrix}$, and natural parameters given by $\boldsymbol{\theta}$ in (6.110).

**Example 6.14 (Bernoulli as Exponential Family)**

Recall the Bernoulli distribution from Example 6.8:

$$ p(x | \mu) = \mu^x (1 - \mu)^{1-x}, \quad x \in \{0, 1\} \quad \text{(6.112)} $$

This can be written in exponential family form:

$$ \begin{aligned} p(x | \mu) &= \exp\left( \log \left( \mu^x (1 - \mu)^{1-x} \right) \right) \quad \text{(6.113a)} \\ &= \exp [x \log \mu + (1 - x) \log(1 - \mu)] \quad \text{(6.113b)} \\ &= \exp [x \log \mu - x \log(1 - \mu) + \log(1 - \mu)] \quad \text{(6.113c)} \\ &= \exp \left[ x \log \frac{\mu}{1-\mu} + \log(1 - \mu) \right] \quad \text{(6.113d)} \end{aligned} $$

The last line (6.113d) can be identified as being in exponential family form (6.107) by observing that:

$$ h(x) = 1 \quad \text{(6.114)} $$

```python
import numpy as np

# --- 6.6.3 Exponential Family Concepts ---

# The general form of an Exponential Family distribution:
# p(x | theta) = h(x) * exp( <theta, phi(x)> - A(theta) )
# Or, more intuitively:
# p(x | theta) propto exp( theta_T * phi(x) )

# --- Example 6.13: Gaussian as Exponential Family ---

print("--- Example 6.13: Gaussian as Exponential Family ---")

# 1. Univariate Gaussian PDF (Standard form)
def gaussian_pdf(x, mu, sigma_sq):
    """
    Calculates the probability density function (PDF) of a univariate Gaussian.
    x: value
    mu: mean
    sigma_sq: variance (sigma^2)
    """
    if sigma_sq <= 0:
        raise ValueError("Variance (sigma_sq) must be positive.")
    
    coeff = 1 / np.sqrt(2 * np.pi * sigma_sq)
    exponent_term = - (x - mu)**2 / (2 * sigma_sq)
    return coeff * np.exp(exponent_term)

# 2. Components for Gaussian in Exponential Family Form
def gaussian_phi(x):
    """Sufficient statistics vector phi(x) for Gaussian."""
    return np.array([x, x**2])

def gaussian_natural_parameters(mu, sigma_sq):
    """Natural parameters theta for Gaussian."""
    if sigma_sq <= 0:
        raise ValueError("Variance (sigma_sq) must be positive.")
    theta1 = mu / sigma_sq
    theta2 = -1 / (2 * sigma_sq)
    return np.array([theta1, theta2])

def gaussian_h_x(x):
    """Base measure h(x) for Gaussian. As per text, can be 1 for proportional form."""
    return 1 

def gaussian_log_partition_function(mu, sigma_sq):
    """
    Log-partition function A(theta) for Gaussian when h(x)=1.
    A(theta) = (mu^2 / (2*sigma_sq)) + 0.5 * log(2 * pi * sigma_sq)
    """
    return (mu**2) / (2 * sigma_sq) + 0.5 * np.log(2 * np.pi * sigma_sq)


# 3. Demonstrate Gaussian (proportional form)
# p(x | theta) propto exp(theta_T * phi(x))
def gaussian_exponential_family_proportional(x, mu, sigma_sq):
    """
    Calculates the proportional part of Gaussian in exponential family form.
    Demonstrates exp(theta_T * phi(x)).
    """
    phi_x = gaussian_phi(x)
    theta = gaussian_natural_parameters(mu, sigma_sq)
    dot_product = np.dot(theta, phi_x)
    return np.exp(dot_product)

# Let's pick some values
mu_val = 2.0
sigma_sq_val = 1.5
x_vals = np.linspace(mu_val - 3*np.sqrt(sigma_sq_val), mu_val + 3*np.sqrt(sigma_sq_val), 10)

print(f"\nGaussian parameters: mu={mu_val}, sigma^2={sigma_sq_val}")
print(f"Natural parameters theta: {gaussian_natural_parameters(mu_val, sigma_sq_val)}")

print("\nx_val | Gaussian PDF | Exp. Family (proportional) | Scaled Exp. Family (for comparison)")
print("-" * 80)
# To compare, we need to normalize the exponential family proportional term.
# The normalization constant is exp(-A(theta)).
# So, h(x) * exp(dot_product - A(theta)) = h(x) * exp(dot_product) * exp(-A(theta))
# = h(x) * exp(dot_product) / exp(A(theta))
norm_const_exp_fam = np.exp(gaussian_log_partition_function(mu_val, sigma_sq_val))

for x in x_vals:
    pdf_val = gaussian_pdf(x, mu_val, sigma_sq_val)
    exp_fam_prop_val = gaussian_exponential_family_proportional(x, mu_val, sigma_sq_val)
    # The crucial part is how to scale:
    # We want to show: p(x | theta) = h(x) * exp( <theta, phi(x)> - A(theta) )
    # Which is:       p(x | theta) = h(x) * exp(<theta, phi(x)>) / exp(A(theta))
    scaled_exp_fam_val = gaussian_h_x(x) * exp_fam_prop_val / norm_const_exp_fam
    
    print(f"{x:5.2f} | {pdf_val:12.6f} | {exp_fam_prop_val:24.6f} | {scaled_exp_fam_val:28.6f}")

# Verify that scaled_exp_fam_val is close to pdf_val
assert np.allclose(gaussian_pdf(x_vals, mu_val, sigma_sq_val), 
                   gaussian_h_x(x_vals) * gaussian_exponential_family_proportional(x_vals, mu_val, sigma_sq_val) / norm_const_exp_fam,
                   rtol=1e-5, atol=1e-8), \
                   "Gaussian reconstruction from exponential family form failed!"
print("\nGaussian successfully demonstrated as an Exponential Family member.")

print("\n" + "=" * 80 + "\n")

# --- Example 6.14: Bernoulli as Exponential Family ---

print("--- Example 6.14: Bernoulli as Exponential Family ---")

# 1. Bernoulli PMF (Standard form)
def bernoulli_pmf(x, mu):
    """
    Calculates the probability mass function (PMF) of a Bernoulli distribution.
    x: outcome (0 or 1)
    mu: probability of x=1
    """
    if not (0 <= mu <= 1):
        raise ValueError("Probability mu must be between 0 and 1.")
    if x not in {0, 1}:
        raise ValueError("Outcome x for Bernoulli must be 0 or 1.")
    return (mu**x) * ((1 - mu)**(1 - x))

# 2. Components for Bernoulli in Exponential Family Form (from 6.113d)
def bernoulli_phi(x):
    """Sufficient statistic phi(x) for Bernoulli."""
    return np.array([x]) # phi(x) = x

def bernoulli_natural_parameter(mu):
    """Natural parameter theta for Bernoulli."""
    # Note: mu cannot be exactly 0 or 1 for log(mu/(1-mu))
    if not (0 < mu < 1): 
        raise ValueError("Probability mu must be strictly between 0 and 1 for natural parameter conversion.")
    return np.array([np.log(mu / (1 - mu))]) # theta = log(mu / (1-mu))

def bernoulli_h_x(x):
    """Base measure h(x) for Bernoulli."""
    return 1 # As per text (6.114)

def bernoulli_log_partition_function(mu):
    """Log-partition function A(theta) for Bernoulli."""
    # A(theta) = -log(1-mu) using the mu parameterization, as per the text.
    # It can also be expressed in terms of theta: A(theta) = log(1 + exp(theta))
    # where theta = log(mu/(1-mu)). Let's use the mu form as it's directly given.
    if not (0 < mu < 1):
         raise ValueError("Probability mu must be strictly between 0 and 1 for log-partition function.")
    return -np.log(1 - mu)


# 3. Demonstrate Bernoulli (proportional form)
# p(x | theta) propto exp(theta_T * phi(x))
def bernoulli_exponential_family_proportional(x, mu):
    """
    Calculates the proportional part of Bernoulli in exponential family form.
    Demonstrates exp(theta_T * phi(x)).
    """
    phi_x = bernoulli_phi(x)
    theta = bernoulli_natural_parameter(mu)
    dot_product = np.dot(theta, phi_x)
    return np.exp(dot_product)

# Let's pick a mu value
mu_val_bernoulli = 0.7

print(f"\nBernoulli parameter: mu={mu_val_bernoulli}")
print(f"Natural parameter theta: {bernoulli_natural_parameter(mu_val_bernoulli)}")

print("\nx_val | Bernoulli PMF | Exp. Family (proportional) | Scaled Exp. Family (for comparison)")
print("-" * 80)

# Calculate normalization constant for the proportional form
# Normalization is exp(-A(theta))
norm_const_exp_fam_bernoulli = np.exp(bernoulli_log_partition_function(mu_val_bernoulli))

for x_b in [0, 1]:
    pmf_val = bernoulli_pmf(x_b, mu_val_bernoulli)
    exp_fam_prop_val = bernoulli_exponential_family_proportional(x_b, mu_val_bernoulli)
    scaled_exp_fam_val = bernoulli_h_x(x_b) * exp_fam_prop_val / norm_const_exp_fam_bernoulli
    
    print(f"{x_b:5.0f} | {pmf_val:13.6f} | {exp_fam_prop_val:24.6f} | {scaled_exp_fam_val:28.6f}")

# Verify reconstruction for Bernoulli
assert np.allclose(bernoulli_pmf(0, mu_val_bernoulli), 
                   bernoulli_h_x(0) * bernoulli_exponential_family_proportional(0, mu_val_bernoulli) / norm_const_exp_fam_bernoulli,
                   rtol=1e-5, atol=1e-8), \
                   "Bernoulli reconstruction for x=0 from exponential family form failed!"
assert np.allclose(bernoulli_pmf(1, mu_val_bernoulli), # Corrected spelling of bernoulli_pmf
                   bernoulli_h_x(1) * bernoulli_exponential_family_proportional(1, mu_val_bernoulli) / norm_const_exp_fam_bernoulli,
                   rtol=1e-5, atol=1e-8), \
                   "Bernoulli reconstruction for x=1 from exponential family form failed!"
print("\nBernoulli successfully demonstrated as an Exponential Family member.")

print("\n" + "=" * 80 + "\n")
```

The canonical conjugate prior has the form:

$$ p(\mu | \alpha, \beta) = \exp \left( \alpha \log \frac{\mu}{1-\mu} + (\beta + \alpha) \log(1 - \mu) - A_c (\boldsymbol{\gamma}) \right) \quad \text{(6.122)} $$

where we defined $\boldsymbol{\gamma} := [\alpha, \beta + \alpha]^\top$ and $h_c (\mu) := \mu/(1 - \mu)$. Equation (6.122) then simplifies to:

$$ p(\mu | \alpha, \beta) = \exp [(\alpha - 1) \log \mu + (\beta - 1) \log(1 - \mu) - A_c (\alpha, \beta)] \quad \text{(6.123)} $$

Putting this in non-exponential family form yields:

$$ p(\mu | \alpha, \beta) \propto \mu^{\alpha-1} (1 - \mu)^{\beta-1} \quad \text{(6.124)} $$

which we identify as the **Beta distribution** (6.98). In example 6.12, we assumed that the Beta distribution is the conjugate prior of the Bernoulli distribution and showed that it was indeed the conjugate prior. In this example, we derived the form of the Beta distribution by looking at the canonical conjugate prior of the Bernoulli distribution in exponential family form.

As mentioned in the previous section, the main motivation for exponential families is that they have finite-dimensional sufficient statistics. Additionally, conjugate distributions are easy to write down, and the conjugate distributions also come from an exponential family. From an inference perspective, maximum likelihood estimation behaves nicely because empirical estimates of sufficient statistics are optimal estimates of the population values of sufficient statistics (recall the mean and covariance of a Gaussian). From an optimization perspective, the log-likelihood function is concave, allowing for efficient optimization approaches to be applied (Chapter 7).

```python
import numpy as np
from scipy.special import gamma # For the Beta function in Beta PDF

# --- Bernoulli Distribution (from previous section, included for context) ---
def bernoulli_pmf(x, mu):
    """
    Calculates the probability mass function (PMF) of a Bernoulli distribution.
    x: outcome (0 or 1)
    mu: probability of x=1
    """
    if not (0 <= mu <= 1):
        raise ValueError("Probability mu must be between 0 and 1.")
    if x not in {0, 1}:
        raise ValueError("Outcome x for Bernoulli must be 0 or 1.")
    return (mu**x) * ((1 - mu)**(1 - x))

def bernoulli_phi(x):
    """Sufficient statistic phi(x) for Bernoulli."""
    return x # For Bernoulli, phi(x) = x

def bernoulli_natural_parameter_theta(mu):
    """Natural parameter theta for Bernoulli."""
    if not (0 < mu < 1):
        # Handle edge cases for log(0) or log(inf) for robustness if needed,
        # but for definition, mu should be strictly between 0 and 1.
        if mu == 0: return -np.inf
        if mu == 1: return np.inf
    return np.log(mu / (1 - mu))

def bernoulli_log_partition_function_A(mu):
    """Log-partition function A(theta) for Bernoulli, in terms of mu."""
    if not (0 < mu < 1):
        raise ValueError("Probability mu must be strictly between 0 and 1.")
    return -np.log(1 - mu)

# --- Beta Distribution ---
def beta_pdf(mu, alpha, beta):
    """
    Calculates the probability density function (PDF) of a Beta distribution.
    mu: value (between 0 and 1)
    alpha, beta: shape parameters (alpha > 0, beta > 0)
    """
    if not (0 <= mu <= 1):
        raise ValueError("mu must be between 0 and 1 for Beta distribution.")
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive.")

    # Beta function B(alpha, beta) = Gamma(alpha) * Gamma(beta) / Gamma(alpha + beta)
    beta_func = gamma(alpha) * gamma(beta) / gamma(alpha + beta)
    
    return (mu**(alpha - 1)) * ((1 - mu)**(beta - 1)) / beta_func

# --- Canonical Conjugate Prior (Derivation and Implementation) ---

print("--- Deriving Beta from Canonical Conjugate Prior of Bernoulli ---")

# The text states the canonical conjugate prior has the form:
# p(mu | alpha, beta) = exp( alpha * log(mu/(1-mu)) + (beta + alpha) * log(1-mu) - Ac(gamma) ) (6.122)
# where gamma = [alpha, beta + alpha]^T
#
# It simplifies to:
# p(mu | alpha, beta) = exp( (alpha - 1) * log(mu) + (beta - 1) * log(1-mu) - Ac(alpha, beta) ) (6.123)

# Let's define the components for the prior in exponential family form
# The sufficient statistics for the prior are log(mu) and log(1-mu)
def prior_phi_mu(mu):
    """Sufficient statistics phi_c(mu) for the prior."""
    if not (0 < mu < 1):
        raise ValueError("mu must be strictly between 0 and 1 for log terms.")
    return np.array([np.log(mu), np.log(1 - mu)])

# The natural parameters for the prior are related to alpha and beta
def prior_natural_parameters_lambda(alpha, beta):
    """Natural parameters lambda for the prior."""
    # From (6.123), comparing to exp(lambda_1 * log(mu) + lambda_2 * log(1-mu))
    lambda1 = alpha - 1
    lambda2 = beta - 1
    return np.array([lambda1, lambda2])

# The log-partition function for the prior A_c(alpha, beta)
def prior_log_partition_function_Ac(alpha, beta):
    """
    Log-partition function Ac(alpha, beta) for the Beta distribution.
    This is log(B(alpha, beta)).
    """
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive for Beta function.")
    return np.log(gamma(alpha)) + np.log(gamma(beta)) - np.log(gamma(alpha + beta))


# Implement Equation 6.123 (Exponential Family form of the prior)
def canonical_conjugate_prior_exp_form(mu, alpha, beta):
    """
    Calculates the canonical conjugate prior (Beta) in its exponential family form (Eq 6.123).
    This function includes the normalization term A_c.
    """
    if not (0 < mu < 1): # mu must be strictly between 0 and 1 for logs
        return 0.0 # PMF is zero outside (0,1)

    log_term_1 = (alpha - 1) * np.log(mu)
    log_term_2 = (beta - 1) * np.log(1 - mu)
    
    Ac_val = prior_log_partition_function_Ac(alpha, beta)
    
    return np.exp(log_term_1 + log_term_2 - Ac_val)

# Implement Equation 6.124 (Proportional form of the prior)
def canonical_conjugate_prior_proportional_form(mu, alpha, beta):
    """
    Calculates the canonical conjugate prior (Beta) in its proportional form (Eq 6.124).
    This function does NOT include the normalization constant.
    """
    if not (0 <= mu <= 1):
        return 0.0
    
    return (mu**(alpha - 1)) * ((1 - mu)**(beta - 1))

# --- Demonstration and Verification ---

# Set example parameters for the Beta distribution
alpha_val = 2.5
beta_val = 3.0

print(f"\nBeta distribution parameters: alpha={alpha_val}, beta={beta_val}")

# Pick some values of mu (probability) to evaluate
mu_values = np.array([0.1, 0.3, 0.5, 0.7, 0.9])

print("\nmu_val | Beta PDF | Prior Exp. Form (6.123) | Prior Proportional Form (6.124)")
print("-" * 100)

# Calculate the normalization constant for the proportional form for comparison
# This is 1 / Beta(alpha, beta)
beta_normalization_const = 1 / (gamma(alpha_val) * gamma(beta_val) / gamma(alpha_val + beta_val))

for mu_val in mu_values:
    beta_val_pdf = beta_pdf(mu_val, alpha_val, beta_val)
    exp_form_val = canonical_conjugate_prior_exp_form(mu_val, alpha_val, beta_val)
    prop_form_val = canonical_conjugate_prior_proportional_form(mu_val, alpha_val, beta_val)
    
    # Scale the proportional form to match the PDF
    scaled_prop_form_val = prop_form_val * beta_normalization_const
    
    print(f"{mu_val:6.3f} | {beta_val_pdf:8.6f} | {exp_form_val:23.6f} | {prop_form_val:32.6f} (Scaled: {scaled_prop_form_val:8.6f})")

# Assertions to confirm the equivalence
for mu_val in mu_values:
    # Check if (6.123) equals Beta PDF
    assert np.isclose(canonical_conjugate_prior_exp_form(mu_val, alpha_val, beta_val),
                      beta_pdf(mu_val, alpha_val, beta_val), rtol=1e-5, atol=1e-8), \
                      f"Eq 6.123 does not match Beta PDF for mu={mu_val}"
    
    # Check if (6.124) proportional form, when scaled, equals Beta PDF
    prop_val = canonical_conjugate_prior_proportional_form(mu_val, alpha_val, beta_val)
    scaled_prop_val = prop_val * beta_normalization_const
    assert np.isclose(scaled_prop_val,
                      beta_pdf(mu_val, alpha_val, beta_val), rtol=1e-5, atol=1e-8), \
                      f"Eq 6.124 (scaled) does not match Beta PDF for mu={mu_val}"

print("\nDemonstration successful: The derived canonical conjugate prior forms match the Beta distribution.")
print("This shows how the Beta distribution naturally arises from the exponential family structure of the Bernoulli distribution's conjugate prior.")
```

```python
import math

# --- Helper functions (replacing numpy/scipy) ---
# A simple definition for pi
MY_PI = 3.141592653589793

# Gamma function (placeholder - real implementation is complex)
# For demonstration purposes, we'll use a very limited conceptual approach
# or acknowledge its necessity.
# A proper Gamma function (e.g., using Lanczos approximation or infinite product)
# is non-trivial to implement from scratch.
# For integer/half-integer values, it's easier: Gamma(n) = (n-1)!
# Since Beta distribution needs Gamma(alpha), Gamma(beta), Gamma(alpha+beta),
# and alpha/beta can be non-integers, a true 'core python' gamma is very hard.
# For the purpose of this demo, I'll provide a factorial-based 'gamma' for integer inputs
# and state that for non-integers, a real library would be needed.

def _factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Factorial is defined for non-negative integers.")
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def _gamma_approx_for_integers(x):
    """
    Conceptual Gamma function approximation for integers (Gamma(n) = (n-1)!).
    This is NOT a general Gamma function.
    Will raise an error for non-integer or non-positive input.
    """
    if x <= 0:
        raise ValueError("Gamma function for Beta requires positive inputs. For non-positive, it's complex.")
    if x == 1:
        return 1.0 # Gamma(1) = 1
    if x == 0.5: # Gamma(0.5) = sqrt(pi)
        return math.sqrt(MY_PI)
    if x > 0 and x == int(x): # For positive integers, Gamma(n) = (n-1)!
        return float(_factorial(int(x) - 1))
    
    # For non-integer positive x, a full Gamma implementation is required,
    # which is computationally very intensive and beyond a simple 'core python' demo.
    # We will conceptually use it but acknowledge this limitation for actual calculation.
    # For the provided Beta example (alpha=2.5, beta=3.0), this simplified gamma will fail.
    # Therefore, the Beta PDF and related calculations will be limited to integer alpha/beta
    # or will be marked as conceptual.
    raise NotImplementedError(f"True Gamma function for x={x} (non-integer) not implemented in core Python.")


# --- Example 6.13: Gaussian as Exponential Family (Core Python) ---

print("--- Example 6.13: Gaussian as Exponential Family (Core Python) ---")

# 1. Univariate Gaussian PDF (Standard form)
def gaussian_pdf_core_python(x, mu, sigma_sq):
    """
    Calculates the PDF of a univariate Gaussian using only built-in math functions.
    x: value
    mu: mean
    sigma_sq: variance (sigma^2)
    """
    if sigma_sq <= 0:
        raise ValueError("Variance (sigma_sq) must be positive.")
    
    coeff = 1 / math.sqrt(2 * MY_PI * sigma_sq)
    exponent_term = - (x - mu)**2 / (2 * sigma_sq)
    return coeff * math.exp(exponent_term)

# 2. Components for Gaussian in Exponential Family Form (Core Python)
# These are just structural; no complex math here
def gaussian_phi_core_python(x):
    """Sufficient statistics vector phi(x) for Gaussian."""
    return [x, x**2] # Represent as a list

def gaussian_natural_parameters_core_python(mu, sigma_sq):
    """Natural parameters theta for Gaussian."""
    if sigma_sq <= 0:
        raise ValueError("Variance (sigma_sq) must be positive.")
    theta1 = mu / sigma_sq
    theta2 = -1 / (2 * sigma_sq)
    return [theta1, theta2]

def gaussian_h_x_core_python(x):
    """Base measure h(x) for Gaussian. As per text, can be 1 for proportional form."""
    return 1 

def gaussian_log_partition_function_core_python(mu, sigma_sq):
    """
    Log-partition function A(theta) for Gaussian when h(x)=1.
    A(theta) = (mu^2 / (2*sigma_sq)) + 0.5 * log(2 * pi * sigma_sq)
    """
    return (mu**2) / (2 * sigma_sq) + 0.5 * math.log(2 * MY_PI * sigma_sq)


# 3. Demonstrate Gaussian (proportional form)
def gaussian_exponential_family_proportional_core_python(x, mu, sigma_sq):
    """
    Calculates the proportional part of Gaussian in exponential family form.
    Demonstrates exp(theta_T * phi(x)).
    """
    phi_x = gaussian_phi_core_python(x)
    theta = gaussian_natural_parameters_core_python(mu, sigma_sq)
    
    # Dot product for lists
    dot_product = theta[0] * phi_x[0] + theta[1] * phi_x[1]
    return math.exp(dot_product)

# Let's pick some values
mu_val = 2.0
sigma_sq_val = 1.5
# For numerical comparison, we need specific points, not arrays
x_test_points = [mu_val - 2*math.sqrt(sigma_sq_val), mu_val, mu_val + 2*math.sqrt(sigma_sq_val)]

print(f"\nGaussian parameters: mu={mu_val}, sigma^2={sigma_sq_val}")
print(f"Natural parameters theta: {gaussian_natural_parameters_core_python(mu_val, sigma_sq_val)}")

print("\nx_val | Gaussian PDF | Exp. Family (proportional) | Scaled Exp. Family (for comparison)")
print("-" * 80)
# To compare, we need to normalize the exponential family proportional term.
norm_const_exp_fam = math.exp(gaussian_log_partition_function_core_python(mu_val, sigma_sq_val))

for x_val in x_test_points:
    pdf_val = gaussian_pdf_core_python(x_val, mu_val, sigma_sq_val)
    exp_fam_prop_val = gaussian_exponential_family_proportional_core_python(x_val, mu_val, sigma_sq_val)
    scaled_exp_fam_val = gaussian_h_x_core_python(x_val) * exp_fam_prop_val / norm_const_exp_fam
    
    print(f"{x_val:5.2f} | {pdf_val:12.6f} | {exp_fam_prop_val:24.6f} | {scaled_exp_fam_val:28.6f}")

# Manual assertion for a single point due to float precision and non-array nature
# This is less robust than numpy.allclose for multiple points.
tolerance = 1e-6
assert abs(gaussian_pdf_core_python(mu_val, mu_val, sigma_sq_val) - \
           (gaussian_h_x_core_python(mu_val) * gaussian_exponential_family_proportional_core_python(mu_val, mu_val, sigma_sq_val) / norm_const_exp_fam)) < tolerance, \
           "Gaussian reconstruction from exponential family form failed for mean!"
print("\nGaussian successfully demonstrated as an Exponential Family member (conceptually).")

print("\n" + "=" * 80 + "\n")

# --- Example 6.14: Bernoulli as Exponential Family (Core Python) ---

print("--- Example 6.14: Bernoulli as Exponential Family (Core Python) ---")

# 1. Bernoulli PMF (Standard form)
def bernoulli_pmf_core_python(x, mu):
    """
    Calculates the probability mass function (PMF) of a Bernoulli distribution.
    x: outcome (0 or 1)
    mu: probability of x=1
    """
    if not (0 <= mu <= 1):
        raise ValueError("Probability mu must be between 0 and 1.")
    if x not in {0, 1}:
        raise ValueError("Outcome x for Bernoulli must be 0 or 1.")
    return (mu**x) * ((1 - mu)**(1 - x))

# 2. Components for Bernoulli in Exponential Family Form
def bernoulli_phi_core_python(x):
    """Sufficient statistic phi(x) for Bernoulli."""
    return x # phi(x) = x

def bernoulli_natural_parameter_core_python(mu):
    """Natural parameter theta for Bernoulli."""
    if not (0 < mu < 1): 
        if mu == 0: return -float('inf')
        if mu == 1: return float('inf')
        raise ValueError("Probability mu must be strictly between 0 and 1 for natural parameter conversion.")
    return math.log(mu / (1 - mu))

def bernoulli_h_x_core_python(x):
    """Base measure h(x) for Bernoulli."""
    return 1 # As per text (6.114)

def bernoulli_log_partition_function_core_python(mu):
    """Log-partition function A(theta) for Bernoulli."""
    if not (0 < mu < 1):
         raise ValueError("Probability mu must be strictly between 0 and 1 for log-partition function.")
    return -math.log(1 - mu)


# 3. Demonstrate Bernoulli (proportional form)
def bernoulli_exponential_family_proportional_core_python(x, mu):
    """
    Calculates the proportional part of Bernoulli in exponential family form.
    Demonstrates exp(theta_T * phi(x)).
    """
    phi_x = bernoulli_phi_core_python(x)
    theta = bernoulli_natural_parameter_core_python(mu)
    
    # Dot product for single values
    dot_product = theta * phi_x # x is scalar, theta is scalar
    return math.exp(dot_product)

# Let's pick a mu value
mu_val_bernoulli = 0.7

print(f"\nBernoulli parameter: mu={mu_val_bernoulli}")
print(f"Natural parameter theta: {bernoulli_natural_parameter_core_python(mu_val_bernoulli)}")

print("\nx_val | Bernoulli PMF | Exp. Family (proportional) | Scaled Exp. Family (for comparison)")
print("-" * 80)

norm_const_exp_fam_bernoulli = math.exp(bernoulli_log_partition_function_core_python(mu_val_bernoulli))

for x_b in [0, 1]:
    pmf_val = bernoulli_pmf_core_python(x_b, mu_val_bernoulli)
    exp_fam_prop_val = bernoulli_exponential_family_proportional_core_python(x_b, mu_val_bernoulli)
    scaled_exp_fam_val = bernoulli_h_x_core_python(x_b) * exp_fam_prop_val / norm_const_exp_fam_bernoulli
    
    print(f"{x_b:5.0f} | {pmf_val:13.6f} | {exp_fam_prop_val:24.6f} | {scaled_exp_fam_val:28.6f}")

# Manual assertion for Bernoulli
assert abs(bernoulli_pmf_core_python(0, mu_val_bernoulli) - \
           (bernoulli_h_x_core_python(0) * bernoulli_exponential_family_proportional_core_python(0, mu_val_bernoulli) / norm_const_exp_fam_bernoulli)) < tolerance, \
           "Bernoulli reconstruction for x=0 from exponential family form failed!"
assert abs(bernoulli_pmf_core_python(1, mu_val_bernoulli) - \
           (bernoulli_h_x_core_python(1) * bernoulli_exponential_family_proportional_core_python(1, mu_val_bernoulli) / norm_const_exp_fam_bernoulli)) < tolerance, \
           "Bernoulli reconstruction for x=1 from exponential family form failed!"
print("\nBernoulli successfully demonstrated as an Exponential Family member (conceptually).")

print("\n" + "=" * 80 + "\n")


# --- Canonical Conjugate Prior (Beta) as Exponential Family (Core Python) ---

print("--- Deriving Beta from Canonical Conjugate Prior of Bernoulli (Core Python) ---")

# Beta distribution PDF (Core Python)
def beta_pdf_core_python(mu, alpha, beta):
    """
    Calculates the PDF of a Beta distribution.
    mu: value (between 0 and 1)
    alpha, beta: shape parameters (alpha > 0, beta > 0)
    
    NOTE: This requires a TRUE Gamma function for non-integer alpha/beta.
    For this core Python demo, we'll try to use _gamma_approx_for_integers.
    If alpha or beta are non-integers, it will raise NotImplementedError.
    """
    if not (0 <= mu <= 1):
        raise ValueError("mu must be between 0 and 1 for Beta distribution.")
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive.")

    try:
        # Beta function B(alpha, beta) = Gamma(alpha) * Gamma(beta) / Gamma(alpha + beta)
        beta_func = _gamma_approx_for_integers(alpha) * _gamma_approx_for_integers(beta) / _gamma_approx_for_integers(alpha + beta)
    except NotImplementedError as e:
        print(f"\nWARNING: Cannot calculate Beta PDF for non-integer alpha/beta without full Gamma function: {e}")
        print("Numerical comparison for Beta PDF will be skipped or may fail for these parameters.")
        return float('nan') # Return Not a Number if Gamma not implemented
    except ValueError as e:
        print(f"\nWARNING: Gamma function error for Beta PDF: {e}")
        return float('nan')

    return (mu**(alpha - 1)) * ((1 - mu)**(beta - 1)) / beta_func

# Components for the prior in exponential family form (Core Python)
def prior_phi_mu_core_python(mu):
    """Sufficient statistics phi_c(mu) for the prior."""
    if not (0 < mu < 1):
        raise ValueError("mu must be strictly between 0 and 1 for log terms.")
    return [math.log(mu), math.log(1 - mu)]

def prior_natural_parameters_lambda_core_python(alpha, beta):
    """Natural parameters lambda for the prior."""
    lambda1 = alpha - 1
    lambda2 = beta - 1
    return [lambda1, lambda2]

def prior_log_partition_function_Ac_core_python(alpha, beta):
    """
    Log-partition function Ac(alpha, beta) for the Beta distribution.
    This is log(B(alpha, beta)).
    
    NOTE: Requires _gamma_approx_for_integers.
    """
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive for Beta function.")
    try:
        return math.log(_gamma_approx_for_integers(alpha)) + \
               math.log(_gamma_approx_for_integers(beta)) - \
               math.log(_gamma_approx_for_integers(alpha + beta))
    except NotImplementedError as e:
        print(f"\nWARNING: Cannot calculate Ac for non-integer alpha/beta without full Gamma function: {e}")
        return float('nan')
    except ValueError as e:
        print(f"\nWARNING: Gamma function error for Ac: {e}")
        return float('nan')

# Implement Equation 6.123 (Exponential Family form of the prior)
def canonical_conjugate_prior_exp_form_core_python(mu, alpha, beta):
    """
    Calculates the canonical conjugate prior (Beta) in its exponential family form (Eq 6.123).
    This function includes the normalization term A_c.
    """
    if not (0 < mu < 1): 
        return 0.0

    log_term_1 = (alpha - 1) * math.log(mu)
    log_term_2 = (beta - 1) * math.log(1 - mu)
    
    Ac_val = prior_log_partition_function_Ac_core_python(alpha, beta)
    
    if math.isnan(Ac_val): # If Ac couldn't be computed
        return float('nan')

    return math.exp(log_term_1 + log_term_2 - Ac_val)

# Implement Equation 6.124 (Proportional form of the prior)
def canonical_conjugate_prior_proportional_form_core_python(mu, alpha, beta):
    """
    Calculates the canonical conjugate prior (Beta) in its proportional form (Eq 6.124).
    This function does NOT include the normalization constant.
    """
    if not (0 <= mu <= 1):
        return 0.0
    
    return (mu**(alpha - 1)) * ((1 - mu)**(beta - 1))

# --- Demonstration and Verification ---

# Set example parameters for the Beta distribution
# IMPORTANT: For this core python version, alpha and beta should be integers
# for the gamma function approximation to work without NotImplementedError.
alpha_val_beta = 2
beta_val_beta = 3

print(f"\nBeta distribution parameters: alpha={alpha_val_beta}, beta={beta_val_beta}")

mu_values_beta = [0.1, 0.3, 0.5, 0.7, 0.9]

print("\nmu_val | Beta PDF | Prior Exp. Form (6.123) | Prior Proportional Form (6.124)")
print("-" * 100)

# Calculate the normalization constant for the proportional form for comparison
try:
    beta_normalization_const_core = 1 / (_gamma_approx_for_integers(alpha_val_beta) * _gamma_approx_for_integers(beta_val_beta) / _gamma_approx_for_integers(alpha_val_beta + beta_val_beta))
except Exception as e:
    print(f"Cannot calculate beta normalization constant: {e}")
    beta_normalization_const_core = float('nan')


for mu_val in mu_values_beta:
    beta_val_pdf = beta_pdf_core_python(mu_val, alpha_val_beta, beta_val_beta)
    exp_form_val = canonical_conjugate_prior_exp_form_core_python(mu_val, alpha_val_beta, beta_val_beta)
    prop_form_val = canonical_conjugate_prior_proportional_form_core_python(mu_val, alpha_val_beta, beta_val_beta)
    
    scaled_prop_form_val = float('nan') # Initialize
    if not math.isnan(beta_normalization_const_core):
        scaled_prop_form_val = prop_form_val * beta_normalization_const_core
    
    print(f"{mu_val:6.3f} | {beta_val_pdf:8.6f} | {exp_form_val:23.6f} | {prop_form_val:32.6f} (Scaled: {scaled_prop_form_val:8.6f})")

# Assertions to confirm the equivalence (only for cases where Gamma is computable)
if not math.isnan(beta_normalization_const_core):
    for mu_val in mu_values_beta:
        if not math.isnan(beta_pdf_core_python(mu_val, alpha_val_beta, beta_val_beta)): # Only assert if PDF is computable
            # Check if (6.123) equals Beta PDF
            assert abs(canonical_conjugate_prior_exp_form_core_python(mu_val, alpha_val_beta, beta_val_beta) - \
                       beta_pdf_core_python(mu_val, alpha_val_beta, beta_val_beta)) < tolerance, \
                       f"Eq 6.123 does not match Beta PDF for mu={mu_val}"
            
            # Check if (6.124) proportional form, when scaled, equals
```

```python
import math

# --- Helper functions (from previous "no numpy" implementation) ---
MY_PI = 3.141592653589793

def _factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Factorial is defined for non-negative integers.")
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def _gamma_approx_for_integers(x):
    """
    Conceptual Gamma function approximation for integers (Gamma(n) = (n-1)!).
    This is NOT a general Gamma function.
    Will raise an error for non-integer or non-positive input.
    """
    if x <= 0:
        raise ValueError("Gamma function for Beta requires positive inputs. For non-positive, it's complex.")
    if x == 1:
        return 1.0 # Gamma(1) = 1
    if x == 0.5: # Gamma(0.5) = sqrt(pi) - special case
        return math.sqrt(MY_PI)
    if x > 0 and x == int(x): # For positive integers, Gamma(n) = (n-1)!
        return float(_factorial(int(x) - 1))
    
    raise NotImplementedError(f"True Gamma function for x={x} (non-integer) not implemented in core Python.")

# --- Gaussian PDF (Core Python) ---
def gaussian_pdf_core_python(x, mu, sigma_sq):
    if sigma_sq <= 0:
        return 0.0 # Invalid variance
    coeff = 1 / math.sqrt(2 * MY_PI * sigma_sq)
    exponent_term = - (x - mu)**2 / (2 * sigma_sq)
    return coeff * math.exp(exponent_term)

# --- Beta PDF (Core Python) ---
def beta_pdf_core_python(mu, alpha, beta):
    if not (0 <= mu <= 1):
        return 0.0
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive.")

    try:
        beta_func = _gamma_approx_for_integers(alpha) * _gamma_approx_for_integers(beta) / _gamma_approx_for_integers(alpha + beta)
    except NotImplementedError:
        print("\nWARNING: Beta PDF cannot be fully calculated for these parameters without a true Gamma function.")
        return float('nan') # Return Not a Number
    except ValueError:
        print("\nWARNING: Gamma function error in Beta PDF calculation.")
        return float('nan')

    numerator = (mu**(alpha - 1)) * ((1 - mu)**(beta - 1))
    return numerator / beta_func

# --- ASCII Plotter ---
def plot_ascii(func, x_min, x_max, title="", num_points=50, height=10, func_args=None):
    """
    Generates a very basic ASCII art plot for a given function.
    func: The function to plot (e.g., gaussian_pdf_core_python)
    x_min, x_max: Range of x-values for the plot
    title: Plot title
    num_points: Number of horizontal "pixels" for the plot
    height: Number of vertical "pixels" for the plot
    func_args: Tuple of arguments to pass to func after x
    """
    if func_args is None:
        func_args = ()

    x_vals = [x_min + i * (x_max - x_min) / (num_points - 1) for i in range(num_points)]
    y_vals = [func(x, *func_args) for x in x_vals]

    # Handle NaN values gracefully if func failed (e.g., Beta with non-integer alpha/beta)
    if any(math.isnan(y) for y in y_vals):
        print(f"\nSkipping ASCII plot for '{title}' due to NaN values (function not fully computable).")
        return

    y_min_val = min(y_vals)
    y_max_val = max(y_vals)

    if y_max_val == y_min_val: # Flat line case
        y_max_val = y_min_val + 1e-6 # Avoid division by zero

    print(f"\n--- {title} ---")
    print(f"X range: [{x_min:.2f}, {x_max:.2f}]")
    print(f"Y range: [{y_min_val:.2f}, {y_max_val:.2f}]\n")

    plot_grid = [[' ' for _ in range(num_points)] for _ in range(height)]

    # Populate the grid
    for i, y in enumerate(y_vals):
        if not math.isnan(y): # Only plot if y is a number
            # Scale y to fit the height
            y_scaled = int((y - y_min_val) / (y_max_val - y_min_val) * (height - 1))
            y_scaled = max(0, min(y_scaled, height - 1)) # Clamp to grid boundaries
            plot_grid[height - 1 - y_scaled][i] = '*' # Plot as '*'

    # Print the grid
    for row in plot_grid:
        print("".join(row))
    print("-" * num_points) # X-axis


# --- Plotting Gaussian ---
print("Attempting to plot Gaussian Distribution (ASCII art)...")
mu_g = 0
sigma_sq_g = 1
sigma_g = math.sqrt(sigma_sq_g)
plot_ascii(gaussian_pdf_core_python, 
           mu_g - 3*sigma_g, 
           mu_g + 3*sigma_g, 
           title=f'Gaussian (mu={mu_g}, sigma^2={sigma_sq_g})', 
           func_args=(mu_g, sigma_sq_g),
           num_points=60, height=15)

# --- Plotting Beta (requires integer alpha/beta for simple Gamma) ---
print("\nAttempting to plot Beta Distribution (ASCII art)...")

# Example 1: Symmetric, bell-shaped (requires integer alpha/beta for this demo)
alpha1, beta1 = 2, 2
plot_ascii(beta_pdf_core_python,
           0.01, 0.99, # Avoid 0 and 1 due to power issues with alpha/beta < 1
           title=f'Beta (alpha={alpha1}, beta={beta1})',
           func_args=(alpha1, beta1),
           num_points=60, height=15)

# Example 2: U-shaped (requires integer alpha/beta for this demo, will fail for 0.5)
# This will likely trigger the NotImplementedError due to non-integer 0.5.
alpha2, beta2 = 1, 1 # This is Uniform distribution, also Beta(1,1)
plot_ascii(beta_pdf_core_python,
           0.01, 0.99,
           title=f'Beta (alpha={alpha2}, beta={beta2})',
           func_args=(alpha2, beta2),
           num_points=60, height=15)

# Example 3: Skewed right (requires integer alpha/beta for this demo)
alpha3, beta3 = 5, 2
plot_ascii(beta_pdf_core_python,
           0.01, 0.99,
           title=f'Beta (alpha={alpha3}, beta={beta3})',
           func_args=(alpha3, beta3),
           num_points=60, height=15)

print("\nASCII art plots are a very limited way to visualize data without proper plotting libraries.")
print("For actual scientific plotting, libraries like matplotlib (which uses numpy) are indispensable.")
```

```python
import numpy as np # Seaborn relies on numpy for data arrays
import matplotlib.pyplot as plt # Seaborn relies on matplotlib for plotting backend
import seaborn as sns
from scipy.special import gamma # Necessary for the Beta function's normalization

# Set a style for better aesthetics
sns.set_theme(style="whitegrid")

# --- Gaussian Distribution (with NumPy for plotting) ---

def gaussian_pdf_np(x, mu, sigma_sq):
    """
    Calculates the probability density function (PDF) of a univariate Gaussian using NumPy.
    x: value (can be a scalar or NumPy array)
    mu: mean
    sigma_sq: variance (sigma^2)
    """
    if sigma_sq <= 0:
        raise ValueError("Variance (sigma_sq) must be positive.")
    
    coeff = 1 / np.sqrt(2 * np.pi * sigma_sq)
    exponent_term = - (x - mu)**2 / (2 * sigma_sq)
    return coeff * np.exp(exponent_term)

# Plotting the Gaussian Distribution
print("Plotting Gaussian Distribution using Seaborn...")
mu_val_gaussian = 0
sigma_sq_val_gaussian = 1
sigma_val_gaussian = np.sqrt(sigma_sq_val_gaussian)

# Generate x values using numpy
x_gaussian = np.linspace(mu_val_gaussian - 4*sigma_val_gaussian, mu_val_gaussian + 4*sigma_val_gaussian, 500)
pdf_gaussian = gaussian_pdf_np(x_gaussian, mu_val_gaussian, sigma_sq_val_gaussian)

plt.figure(figsize=(8, 5)) # Matplotlib figure
sns.lineplot(x=x_gaussian, y=pdf_gaussian, color='blue', label=f'Gaussian ($\mu$={mu_val_gaussian}, $\sigma^2$={sigma_sq_val_gaussian})')
plt.title('Univariate Gaussian Distribution (PDF) via Seaborn') # Matplotlib title
plt.xlabel('x') # Matplotlib xlabel
plt.ylabel('Probability Density') # Matplotlib ylabel
plt.legend() # Matplotlib legend
plt.grid(True) # Matplotlib grid
plt.show() # Matplotlib show

# --- Beta Distribution (with NumPy and SciPy for plotting) ---

def beta_pdf_np(mu, alpha, beta):
    """
    Calculates the probability density function (PDF) of a Beta distribution using NumPy and SciPy.
    mu: value (can be a scalar or NumPy array, between 0 and 1)
    alpha, beta: shape parameters (alpha > 0, beta > 0)
    """
    if np.any((mu < 0) | (mu > 1)):
        raise ValueError("mu values must be between 0 and 1 for Beta distribution.")
    if alpha <= 0 or beta <= 0:
        raise ValueError("Alpha and Beta parameters must be positive.")

    # Beta function B(alpha, beta) = Gamma(alpha) * Gamma(beta) / Gamma(alpha + beta)
    beta_func = gamma(alpha) * gamma(beta) / gamma(alpha + beta)
    
    return (mu**(alpha - 1)) * ((1 - mu)**(beta - 1)) / beta_func

# Plotting the Beta Distribution with different parameters
print("\nPlotting Beta Distribution for various parameters using Seaborn...")

# Generate mu values using numpy
mu_beta = np.linspace(0.001, 0.999, 500) # Avoid exact 0 or 1 to prevent issues if alpha/beta < 1

plt.figure(figsize=(10, 6)) # Matplotlib figure

# Example 1: Symmetric, bell-shaped
alpha1, beta1 = 2, 2
pdf1 = beta_pdf_np(mu_beta, alpha1, beta1)
sns.lineplot(x=mu_beta, y=pdf1, label=f'Beta($\\alpha$={alpha1}, $\\beta$={beta1})', color='green')

# Example 2: U-shaped
alpha2, beta2 = 0.5, 0.5
pdf2 = beta_pdf_np(mu_beta, alpha2, beta2)
sns.lineplot(x=mu_beta, y=pdf2, label=f'Beta($\\alpha$={alpha2}, $\\beta$={beta2})', color='red')

# Example 3: Skewed left
alpha3, beta3 = 5, 1
pdf3 = beta_pdf_np(mu_beta, alpha3, beta3)
sns.lineplot(x=mu_beta, y=pdf3, label=f'Beta($\\alpha$={alpha3}, $\\beta$={beta3})', color='purple')

# Example 4: Peaked towards 1
alpha4, beta4 = 1, 5
pdf4 = beta_pdf_np(mu_beta, alpha4, beta4)
sns.lineplot(x=mu_beta, y=pdf4, label=f'Beta($\\alpha$={alpha4}, $\\beta$={beta4})', color='orange')


plt.title('Beta Distribution (PDF) for Various Parameters via Seaborn') # Matplotlib title
plt.xlabel('$\mu$') # Matplotlib xlabel
plt.ylabel('Probability Density') # Matplotlib ylabel
plt.legend(title="Parameters") # Matplotlib legend
plt.grid(True) # Matplotlib grid
plt.show() # Matplotlib show

print("\nGraphs displayed using Seaborn (which requires NumPy and Matplotlib).")
```

### 6.7 Change of Variables/Inverse Transform

It may seem that there are very many known distributions, but in reality the set of distributions for which we have names is quite limited. Therefore, it is often useful to understand how transformed random variables are distributed. For example, assuming that $X$ is a random variable distributed according to the univariate normal distribution $\mathcal{N}(0, 1)$, what is the distribution of $X^2$? Another example, which is quite common in machine learning, is, given that $X_1$ and $X_2$ are univariate standard normal, what is the distribution of $\frac{1}{2}(X_1 + X_2)$?

One option to work out the distribution of $\frac{1}{2}(X_1 + X_2)$ is to calculate the mean and variance of $X_1$ and $X_2$ and then combine them. As we saw in Section 6.4.4, we can calculate the mean and variance of resulting random variables when we consider affine transformations of random variables. However, we may not be able to obtain the functional form of the distribution under transformations. Furthermore, we may be interested in nonlinear transformations of random variables for which closed-form expressions are not readily available.

**Remark (Notation).** In this section, we will be explicit about random variables and the values they take. Hence, recall that we use capital letters $X, Y$ to denote random variables and small letters $x, y$ to denote the values in the target space $\mathcal{T}$ that the random variables take. We will explicitly write pmfs of discrete random variables $X$ as $P(X = x)$. For continuous random variables $X$ (Section 6.2.2), the pdf is written as $f(x)$ and the cdf is written as $F_X(x)$. $\diamondsuit$

We will look at two approaches for obtaining distributions of transformations of random variables: a direct approach using the definition of a cumulative distribution function and a change-of-variable approach that uses the chain rule of calculus (Section 5.2.2). The change-of-variable approach is widely used because it provides a “recipe” for attempting to compute the resulting distribution due to a transformation. Moment generating functions can also be used to study transformations of random variables (Casella and Berger, 2002, chapter 2). We will explain the techniques for univariate random variables, and will only briefly provide the results for the general case of multivariate random variables.

Transformations of discrete random variables can be understood directly. Suppose that there is a discrete random variable $X$ with pmf $P(X = x)$ (Section 6.2.1), and an invertible function $U(x)$. Consider the transformed random variable $Y := U(X)$, with pmf $P(Y = y)$. Then:

$$ P(Y = y) = P(U(X) = y) \quad \text{transformation of interest (6.125a)} $$
$$ = P(X = U^{-1}(y)) \quad \text{inverse (6.125b)} $$

where we can observe that $x = U^{-1}(y)$. Therefore, for discrete random variables, transformations directly change the individual events (with the probabilities appropriately transformed).

### 6.7.1 Distribution Function Technique

The distribution function technique goes back to first principles, and uses the definition of a cdf $F_X(x) = P(X \leqslant x)$ and the fact that its differential is the pdf $f(x)$ (Wasserman, 2004, chapter 2). For a random variable $X$ and a function $U$, we find the pdf of the random variable $Y := U(X)$ by:

1.  Finding the cdf: $F_Y(y) = P(Y \leqslant y)$ (6.126)
2.  Differentiating the cdf $F_Y(y)$ to get the pdf $f(y)$.

    $$ f(y) = \frac{d}{dy} F_Y(y) \quad \text{(6.127)} $$

We also need to keep in mind that the domain of the random variable may have changed due to the transformation by $U$.

**Example 6.16**

Let $X$ be a continuous random variable with probability density function on $0 \leqslant x \leqslant 1$:

$$ f(x) = 3x^2 \quad \text{(6.128)} $$

We are interested in finding the pdf of $Y = X^2$. The function $f$ is an increasing function of $x$, and therefore the resulting value of $y$ lies in the interval $[0, 1]$. We obtain:

$$ \begin{aligned} F_Y(y) &= P(Y \leqslant y) && \text{definition of cdf (6.129a)} \\ &= P(X^2 \leqslant y) && \text{transformation of interest (6.129b)} \\ &= P(X \leqslant y^{1/2}) && \text{inverse (6.129c)} \\ &= F_X(y^{1/2}) && \text{definition of cdf (6.129d)} \\ &= \int_0^{y^{1/2}} 3t^2 dt && \text{cdf as a definite integral (6.129e)} \\ &= \left[ t^3 \right]_{t=0}^{t=y^{1/2}} && \text{result of integration (6.129f)} \\ &= (y^{1/2})^3 \\ &= y^{3/2}, \quad 0 \leqslant y \leqslant 1. && \text{(6.129g)} \end{aligned} $$

Therefore, the cdf of $Y$ is:

$$ F_Y(y) = y^{3/2} \quad \text{(6.130)} $$

for $0 \leqslant y \leqslant 1$. To obtain the pdf, we differentiate the cdf:

$$ f(y) = \frac{d}{dy} F_Y(y) = \frac{3}{2} y^{1/2} \quad \text{(6.131)} $$

for $0 \leqslant y \leqslant 1$. In Example 6.16, we considered a strictly monotonically increasing function $f(x) = 3x^2$. This means that we could compute an inverse function. In general, we require that the function of interest $y = U(x)$ has an inverse $x = U^{-1}(y)$.

A useful result can be obtained by considering the cumulative distribution function $F_X(x)$ of a random variable $X$, and using it as the transformation $U(x)$. This leads to the following theorem.

**Theorem 6.15.** [Theorem 2.1.10 in Casella and Berger (2002)] Let $X$ be a continuous random variable with a strictly monotonic cumulative distribution function $F_X(x)$. Then the random variable $Y$ defined as $Y := F_X(X)$ (6.132) has a uniform distribution.

Theorem 6.15 is known as the **probability integral transform**, and it is used to derive algorithms for sampling from distributions by transforming the result of sampling from a uniform random variable (Bishop, 2006). The algorithm works by first generating a sample from a uniform distribution, then transforming it by the inverse cdf (assuming this is available) to obtain a sample from the desired distribution. The probability integral transform is also used for hypothesis testing whether a sample comes from a particular distribution (Lehmann and Romano, 2005). The idea that the output of a cdf gives a uniform distribution also forms the basis of copulas (Nelsen, 2006).

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# --- Example 6.16: Transformation of a Continuous Random Variable ---

# Given PDF of X: f_X(x) = 3x^2 for 0 <= x <= 1
def f_X(x):
    """PDF of random variable X."""
    if isinstance(x, (list, np.ndarray)):
        return np.where((x >= 0) & (x <= 1), 3 * x**2, 0.0)
    else:
        return 3 * x**2 if 0 <= x <= 1 else 0.0

# Given CDF of X: F_X(x) = x^3 for 0 <= x <= 1 (derived from integral of f_X(x))
def F_X(x):
    """CDF of random variable X."""
    if isinstance(x, (list, np.ndarray)):
        return np.where(x < 0, 0.0, np.where(x > 1, 1.0, x**3))
    else:
        if x < 0: return 0.0
        if x > 1: return 1.0
        return x**3

# Transformation function: Y = U(X) = X^2
def U(x):
    """Transformation function Y = X^2."""
    return x**2

# Inverse transformation: X = U_inv(Y) = sqrt(Y)
def U_inv(y):
    """Inverse transformation X = sqrt(Y)."""
    return np.sqrt(y)

# 1. Find the CDF of Y: F_Y(y) = P(Y <= y)
def F_Y(y):
    """CDF of transformed random variable Y = X^2."""
    if isinstance(y, (list, np.ndarray)):
        return np.where(y < 0, 0.0, np.where(y > 1, 1.0, F_X(np.sqrt(y))))
    else:
        if y < 0: return 0.0
        if y > 1: return 1.0
        return F_X(math.sqrt(y))

# 2. Differentiate the CDF F_Y(y) to get the PDF f_Y(y)
def f_Y(y):
    """PDF of transformed random variable Y = X^2."""
    if isinstance(y, (list, np.ndarray)):
        return np.where((y >= 0) & (y <= 1), 1.5 * np.sqrt(y), 0.0)
    else:
        return 1.5 * math.sqrt(y) if 0 <= y <= 1 else 0.0

# --- Visualization using Matplotlib ---

print("Plotting PDFs and CDFs for X and Y=X^2...")

# Generate x-values for plotting X's distributions
x_vals = np.linspace(-0.5, 1.5, 400)
fx_vals = f_X(x_vals)
Fx_vals = F_X(x_vals)

# Generate y-values for plotting Y's distributions
y_vals = np.linspace(-0.5, 1.5, 400)
fy_vals = f_Y(y_vals)
Fy_vals = F_Y(y_vals)

# Create plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of X vs. Transformed Variable Y = X^2 (Example 6.16)', fontsize=16)

# Plot f_X(x)
axes[0, 0].plot(x_vals, fx_vals, color='blue')
# CORRECTED LINE HERE: Using \leq instead of \le and escaping the backslash
axes[0, 0].set_title('$f_X(x) = 3x^2$ (for $0 \\leq x \\leq 1$)')
axes[0, 0].set_xlabel('$x$')
axes[0, 0].set_ylabel('Probability Density')
axes[0, 0].grid(True)
axes[0, 0].set_ylim(bottom=0)

# Plot F_X(x)
axes[0, 1].plot(x_vals, Fx_vals, color='green')
# CORRECTED LINE HERE: Using \leq instead of \le and escaping the backslash
axes[0, 1].set_title('$F_X(x) = x^3$ (for $0 \\leq x \\leq 1$)')
axes[0, 1].set_xlabel('$x$')
axes[0, 1].set_ylabel('Cumulative Probability')
axes[0, 1].grid(True)
axes[0, 1].set_ylim(0, 1.1)

# Plot f_Y(y)
axes[1, 0].plot(y_vals, fy_vals, color='red')
# CORRECTED LINE HERE: Using \leq instead of \le and escaping the backslash
axes[1, 0].set_title('$f_Y(y) = \\frac{3}{2}y^{1/2}$ (for $0 \\leq y \\leq 1$)')
axes[1, 0].set_xlabel('$y$')
axes[1, 0].set_ylabel('Probability Density')
axes[1, 0].grid(True)
axes[1, 0].set_ylim(bottom=0)

# Plot F_Y(y)
axes[1, 1].plot(y_vals, Fy_vals, color='purple')
# CORRECTED LINE HERE: Using \leq instead of \le and escaping the backslash
axes[1, 1].set_title('$F_Y(y) = y^{3/2}$ (for $0 \\leq y \\leq 1$)')
axes[1, 1].set_xlabel('$y$')
axes[1, 1].set_ylabel('Cumulative Probability')
axes[1, 1].grid(True)
axes[1, 1].set_ylim(0, 1.1)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("\nDemonstration complete: Plots show the original and transformed distributions.")
print("The change of variables technique was applied to derive Y's distribution.")

# --- Optional: Probability Integral Transform Demonstration (Conceptual) ---
print("\n--- Conceptual: Probability Integral Transform ---")
print("Theorem 6.15: If Y = F_X(X), then Y ~ Uniform(0,1)")

try:
    p_uniform_sample = 0.5
    x_sample = p_uniform_sample**(1/3)
    
    y_transform_result = F_X(x_sample)
    
    print(f"Given a uniform sample p = {p_uniform_sample:.2f}")
    print(f"X sample (from F_X_inv(p)) = {x_sample:.4f}")
    print(f"Transformed Y = F_X(X) = {y_transform_result:.4f}")
    print(f"This value ({y_transform_result:.4f}) should ideally be close to the original uniform sample ({p_uniform_sample:.2f}).")
    print("This demonstrates that transforming a random variable X by its own CDF F_X(X) yields a uniformly distributed variable.")
    print("(Note: For a full statistical demonstration, a large number of samples would be needed.)")

except Exception as e:
    print(f"Could not demonstrate Probability Integral Transform conceptually: {e}")
```

### Change of Variables

The distribution function technique in Section 6.7.1 is derived from first principles, based on the definitions of cdfs and using properties of inverses, differentiation, and integration. This argument from first principles relies on two facts:

1.  We can transform the cdf of $Y$ into an expression that is a cdf of $X$.
2.  We can differentiate the cdf to obtain the pdf.

Let us break down the reasoning step by step, with the goal of understanding the more general change-of-variables approach in Theorem 6.16.

**Remark.** The name “change of variables” comes from the idea of changing the variable of integration when faced with a difficult integral. Change of variables in probability relies on the change-of-variables method in calculus (Tandra, 2014). For univariate functions, we use the substitution rule of integration:

$$ \int f(g(x))g'(x)dx = \int f(u)du, \quad \text{where } u = g(x) \quad \text{(6.133)} $$

The derivation of this rule is based on the chain rule of calculus (5.32) and by applying twice the fundamental theorem of calculus. The fundamental theorem of calculus formalizes the fact that integration and differentiation are somehow “inverses” of each other. An intuitive understanding of the rule can be obtained by thinking (loosely) about small changes (differentials) to the equation $u = g(x)$, that is by considering $\Delta u = g'(x)\Delta x$ as a differential of $u = g(x)$. By substituting $u = g(x)$, the argument inside the integral on the right-hand side of (6.133) becomes $f(g(x))$. By pretending that the term $du$ can be approximated by $du \approx \Delta u = g'(x)\Delta x$, and that $dx \approx \Delta x$, we obtain (6.133). $\diamondsuit$

Consider a univariate random variable $X$, and an invertible function $U$, which gives us another random variable $Y = U(X)$. We assume that random variable $X$ has states $x \in [a, b]$. By the definition of the cdf, we have:

$$ F_Y(y) = P(Y \leqslant y) \quad \text{(6.134)} $$

We are interested in a function $U$ of the random variable:

$$ P(Y \leqslant y) = P(U(X) \leqslant y) \quad \text{(6.135)} $$

where we assume that the function $U$ is invertible. An invertible function on an interval is either strictly increasing or strictly decreasing. In the case that $U$ is strictly increasing, then its inverse $U^{-1}$ is also strictly increasing. By applying the inverse $U^{-1}$ to the arguments of $P(U(X) \leqslant y)$, we obtain:

$$ P(U(X) \leqslant y) = P(U^{-1}(U(X)) \leqslant U^{-1}(y)) = P(X \leqslant U^{-1}(y)) \quad \text{(6.136)} $$

The right-most term in (6.136) is an expression of the cdf of $X$. Recall the definition of the cdf in terms of the pdf:

$$ P(X \leqslant U^{-1}(y)) = \int_a^{U^{-1}(y)} f(x)dx \quad \text{(6.137)} $$

Now we have an expression of the cdf of $Y$ in terms of $x$:

$$ F_Y(y) = \int_a^{U^{-1}(y)} f(x)dx \quad \text{(6.138)} $$

To obtain the pdf, we differentiate (6.138) with respect to $y$:

$$ f(y) = \frac{d}{dy} F_Y(y) = \frac{d}{dy} \int_a^{U^{-1}(y)} f(x)dx \quad \text{(6.139)} $$

Note that the integral on the right-hand side is with respect to $x$, but we need an integral with respect to $y$ because we are differentiating with respect to $y$. In particular, we use (6.133) to get the substitution:

$$ \int f(U^{-1}(y)){U^{-1}}'(y)dy = \int f(x)dx \quad \text{where } x = U^{-1}(y) \quad \text{(6.140)} $$

Using (6.140) on the right-hand side of (6.139) gives us:

$$ f(y) = \frac{d}{dy} \int_a^{U^{-1}(y)} f_x(U^{-1}(y)){U^{-1}}'(y)dy \quad \text{(6.141)} $$

We then recall that differentiation is a linear operator and we use the subscript $x$ to remind ourselves that $f_x(U^{-1}(y))$ is a function of $x$ and not $y$. Invoking the fundamental theorem of calculus again gives us:

$$ f(y) = f_x(U^{-1}(y)) \cdot {U^{-1}}'(y) \quad \text{(6.142)} $$

Recall that we assumed that $U$ is a strictly increasing function. For decreasing functions, it turns out that we have a negative sign when we follow the same derivation. We introduce the **absolute value** of the differential to have the same expression for both increasing and decreasing $U$:

$$ f(y) = f_x(U^{-1}(y)) \cdot \left| \frac{d}{dy} U^{-1}(y) \right| \quad \text{(6.143)} $$

###  Change of Variables (Continued)

This is called the **change-of-variable technique**. The term $\left| \frac{d}{dy} U^{-1}(y) \right|$ in change-of-variable technique (6.143) measures how much a unit volume changes when applying $U$ (see also the definition of the **Jacobian** in Section 5.3).

**Remark.** In comparison to the discrete case in (6.125b), we have an additional factor $\left| \frac{d}{dy} U^{-1}(y) \right|$. The continuous case requires more care because $P(Y = y) = 0$ for all $y$. The probability density function $f(y)$ does not have a description as a probability of an event involving $y$. $\diamondsuit$

So far in this section, we've been studying univariate change of variables. The case for **multivariate random variables** is analogous, but it's complicated by the fact that the absolute value can't be used for multivariate functions. Instead, we use the **determinant of the Jacobian matrix**. Recall from (5.58) that the Jacobian is a matrix of partial derivatives, and that the existence of a nonzero determinant shows that we can invert the Jacobian. Recall the discussion in Section 4.1 that the determinant arises because our differentials (cubes of volume) are transformed into parallelepipeds by the Jacobian.

Let's summarize the preceding discussion in the following theorem, which gives us a recipe for multivariate change of variables.

**Theorem 6.16.** [Theorem 17.2 in Billingsley (1995)]
Let $f_X(\mathbf{x})$ be the value of the probability density of the multivariate continuous random variable $\mathbf{X}$. If the vector-valued function $\mathbf{y} = U(\mathbf{x})$ is differentiable and invertible for all values within the domain of $\mathbf{x}$, then for corresponding values of $\mathbf{y}$, the probability density of $\mathbf{Y} = U(\mathbf{X})$ is given by:

$$f_Y(\mathbf{y}) = f_X(U^{-1}(\mathbf{y})) \cdot \left| \det \left( \frac{\partial}{\partial \mathbf{y}} U^{-1}(\mathbf{y}) \right) \right| \quad \text{(6.144)}$$

The theorem looks intimidating at first glance, but the key point is that a change of variable of a multivariate random variable follows the procedure of the univariate change of variable. First, we need to work out the **inverse transform**, and substitute that into the density of $\mathbf{x}$. Then we calculate the **determinant of the Jacobian** and multiply the result. The following example illustrates the case of a bivariate random variable.

**Example 6.17**

Consider a bivariate random variable $\mathbf{X}$ with states $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ and probability density function:

$$f_X\left( \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \right) = \frac{1}{2\pi} \exp \left( -\frac{1}{2} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}^\top \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \right) \quad \text{(6.145)}$$

We use the change-of-variable technique from Theorem 6.16 to derive the...

### Effect of a Linear Transformation (Section 2.7)

...of the random variable. Consider a matrix $\mathbf{A} \in \mathbb{R}^{2 \times 2}$ defined as:

$$ \mathbf{A} = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \quad \text{(6.146)} $$

We are interested in finding the probability density function of the transformed bivariate random variable $\mathbf{Y}$ with states $\mathbf{y} = \mathbf{Ax}$. Recall that for change of variables we require the inverse transformation of $\mathbf{x}$ as a function of $\mathbf{y}$. Since we consider linear transformations, the inverse transformation is given by the matrix inverse (see Section 2.2.2). For $2 \times 2$ matrices, we can explicitly write out the formula, given by:

$$ \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \mathbf{A}^{-1} \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} \quad \text{(6.147)} $$

Observe that $ad - bc$ is the determinant (Section 4.1) of $\mathbf{A}$. The corresponding probability density function is given by:

$$ f(\mathbf{x}) = f(\mathbf{A}^{-1}\mathbf{y}) = \frac{1}{2\pi} \exp \left( -\frac{1}{2} \mathbf{y}^\top \mathbf{A}^{-\top} \mathbf{A}^{-1} \mathbf{y} \right) \quad \text{(6.148)} $$

The partial derivative of a matrix times a vector with respect to the vector is the matrix itself (Section 5.5), and therefore:

$$ \frac{\partial}{\partial \mathbf{y}} \mathbf{A}^{-1}\mathbf{y} = \mathbf{A}^{-1} \quad \text{(6.149)} $$

Recall from Section 4.1 that the determinant of the inverse is the inverse of the determinant so that the determinant of the Jacobian matrix is:

$$ \left| \det \left( \frac{\partial}{\partial \mathbf{y}} \mathbf{A}^{-1}\mathbf{y} \right) \right| = \left| \det(\mathbf{A}^{-1}) \right| = \left| \frac{1}{\det(\mathbf{A})} \right| = \frac{1}{|ad - bc|} \quad \text{(6.150)} $$

We are now able to apply the change-of-variable formula from Theorem 6.16 by multiplying (6.148) with (6.150), which yields:

$$ \begin{aligned} f_Y(\mathbf{y}) &= f_X(U^{-1}(\mathbf{y})) \left| \det \left( \frac{\partial}{\partial \mathbf{y}} U^{-1}(\mathbf{y}) \right) \right| \quad \text{(6.151a)} \\ &= \frac{1}{2\pi} \exp \left( -\frac{1}{2} \mathbf{y}^\top \mathbf{A}^{-\top} \mathbf{A}^{-1} \mathbf{y} \right) \frac{1}{|ad - bc|} \quad \text{(6.151b)} \end{aligned} $$

While Example 6.17 is based on a bivariate random variable, which allows us to easily compute the matrix inverse, the preceding relation holds for higher dimensions.

**Remark.** We saw in Section 6.5 that the density $f(\mathbf{x})$ in (6.148) is actually the standard Gaussian distribution, and the transformed density $f(\mathbf{y})$ is a bivariate Gaussian with covariance $\mathbf{\Sigma} = \mathbf{A}\mathbf{A}^\top$. $\diamondsuit$

```python
import math

# --- 1. Basic 2x2 Matrix Operations (Core Python) ---

def multiply_matrix_vector_2x2(matrix, vector):
    """Multiplies a 2x2 matrix by a 2x1 vector."""
    if len(matrix) != 2 or len(matrix[0]) != 2 or len(vector) != 2:
        raise ValueError("Matrix must be 2x2 and vector 2x1.")
    
    result = [0, 0]
    result[0] = matrix[0][0] * vector[0] + matrix[0][1] * vector[1]
    result[1] = matrix[1][0] * vector[0] + matrix[1][1] * vector[1]
    return result

def get_determinant_2x2(matrix):
    """Calculates the determinant of a 2x2 matrix."""
    if len(matrix) != 2 or len(matrix[0]) != 2:
        raise ValueError("Matrix must be 2x2.")
    a, b = matrix[0]
    c, d = matrix[1]
    return a * d - b * c

def get_inverse_2x2(matrix):
    """Calculates the inverse of a 2x2 matrix."""
    det = get_determinant_2x2(matrix)
    if det == 0:
        raise ValueError("Matrix is singular and cannot be inverted.")
    
    a, b = matrix[0]
    c, d = matrix[1]
    
    inv_det = 1.0 / det
    
    inv_matrix = [
        [d * inv_det, -b * inv_det],
        [-c * inv_det, a * inv_det]
    ]
    return inv_matrix

def transpose_matrix_2x2(matrix):
    """Transposes a 2x2 matrix."""
    if len(matrix) != 2 or len(matrix[0]) != 2:
        raise ValueError("Matrix must be 2x2.")
    return [
        [matrix[0][0], matrix[1][0]],
        [matrix[0][1], matrix[1][1]]
    ]

def dot_product_vectors(v1, v2):
    """Calculates the dot product of two 2x1 vectors (v1.T @ v2)."""
    if len(v1) != 2 or len(v2) != 2:
        raise ValueError("Vectors must be 2x1.")
    return v1[0] * v2[0] + v1[1] * v2[1]

# --- 2. Bivariate Standard Gaussian PDF for X (Equation 6.145) ---

def fx_standard_bivariate_gaussian(x_vector):
    """
    Calculates the PDF of a standard bivariate Gaussian distribution for X.
    f(x) = (1/(2*pi)) * exp(-0.5 * x_vector.T @ x_vector)
    x_vector: a list [x1, x2]
    """
    if len(x_vector) != 2:
        raise ValueError("x_vector must be a 2x1 vector.")
    
    # x_vector.T @ x_vector = x1^2 + x2^2
    quadratic_form = dot_product_vectors(x_vector, x_vector)
    
    coeff = 1.0 / (2 * math.pi)
    exponent = -0.5 * quadratic_form
    
    return coeff * math.exp(exponent)

# --- 3. Transformation Y = A @ X ---
# This is a conceptual function as Y is the target for which we want PDF.

# --- 4. Inverse Transformation X = A_inv @ Y ---
# Handled by get_inverse_2x2 and multiply_matrix_vector_2x2

# --- 5. Jacobian Determinant |det(d(A_inv @ y)/dy)| ---
# For a linear transformation y = A @ x, x = A_inv @ y
# The partial derivative of A_inv @ y with respect to y is simply A_inv.
# So, the Jacobian determinant is |det(A_inv)| = |1 / det(A)|.

def jacobian_determinant_linear_transform(A_matrix):
    """
    Calculates |det(A_inv)| for a linear transformation Y=AX.
    A_matrix: the 2x2 transformation matrix A.
    """
    det_A = get_determinant_2x2(A_matrix)
    if det_A == 0:
        raise ValueError("Transformation matrix A is singular; Jacobian determinant is zero.")
    return 1.0 / abs(det_A) # Absolute value for the determinant

# --- 6. Final PDF for Y using Change of Variables (Equation 6.151b) ---

def fy_transformed_gaussian(y_vector, A_matrix):
    """
    Calculates the PDF of the transformed random variable Y = A @ X,
    where X is a standard bivariate Gaussian, using the change-of-variables formula.
    f_Y(y) = f_X(A_inv @ y) * |det(A_inv)|
    y_vector: a list [y1, y2] representing the state of Y.
    A_matrix: the 2x2 transformation matrix A.
    """
    if len(y_vector) != 2:
        raise ValueError("y_vector must be a 2x1 vector.")
    
    # Step 1: Calculate the inverse transformation x = A_inv @ y
    A_inv = get_inverse_2x2(A_matrix)
    x_from_y = multiply_matrix_vector_2x2(A_inv, y_vector)
    
    # Step 2: Evaluate the original PDF at the inverse transformed point f_X(A_inv @ y)
    fx_at_x_from_y = fx_standard_bivariate_gaussian(x_from_y)
    
    # Step 3: Calculate the Jacobian determinant |det(A_inv)|
    det_jacobian = jacobian_determinant_linear_transform(A_matrix)
    
    # Step 4: Multiply to get f_Y(y)
    return fx_at_x_from_y * det_jacobian

# --- Example Usage (from Example 6.17 text) ---

print("--- Demonstrating Bivariate Change of Variables (Core Python) ---")

# Define the transformation matrix A (Example 6.17 doesn't give specific values for a,b,c,d,
# so we'll use a simple invertible matrix for demonstration)
A_example = [
    [2.0, 1.0],
    [1.0, 2.0]
]

print(f"\nTransformation Matrix A: {A_example}")
try:
    det_A_example = get_determinant_2x2(A_example)
    print(f"Determinant of A: {det_A_example}")
    A_inv_example = get_inverse_2x2(A_example)
    print(f"Inverse of A: {A_inv_example}")
    jacobian_det_val = jacobian_determinant_linear_transform(A_example)
    print(f"Jacobian determinant |det(A_inv)|: {jacobian_det_val}")

    # Let's evaluate the PDF at a specific y_vector
    y_test_point = [0.5, 0.5]
    print(f"\nEvaluating f_Y(y) at y = {y_test_point}")

    pdf_y_val = fy_transformed_gaussian(y_test_point, A_example)
    print(f"f_Y({y_test_point}) = {pdf_y_val:.6f}")

    # --- Verification (Conceptual) ---
    # For verification, we can calculate the covariance matrix Sigma = A @ A.T
    # and then use the general multivariate Gaussian PDF formula for N(0, Sigma).
    # However, implementing general matrix multiplication and inverse for higher
    # dimensions than 2x2 without numpy is too complex for this demo.
    # We can only conceptually show for the 2x2 case as described in the text.

    # Calculate A @ A.T manually for 2x2
    A_T_example = transpose_matrix_2x2(A_example)
    
    # Manual A @ A.T for 2x2
    sigma_xx = A_example[0][0] * A_T_example[0][0] + A_example[0][1] * A_T_example[1][0]
    sigma_xy = A_example[0][0] * A_T_example[0][1] + A_example[0][1] * A_T_example[1][1]
    sigma_yx = A_example[1][0] * A_T_example[0][0] + A_example[1][1] * A_T_example[1][0]
    sigma_yy = A_example[1][0] * A_T_example[0][1] + A_example[1][1] * A_T_example[1][1]
    
    Covariance_Sigma = [
        [sigma_xx, sigma_xy],
        [sigma_yx, sigma_yy]
    ]
    print(f"\nCalculated Covariance Sigma (A @ A.T): {Covariance_Sigma}")

    # The PDF of Y is N(0, Sigma)
    # f_Y(y) = (1 / (2*pi * sqrt(det(Sigma)))) * exp(-0.5 * y.T @ Sigma_inv @ y)
    
    # Calculate det(Sigma) and Sigma_inv
    det_Sigma = get_determinant_2x2(Covariance_Sigma)
    Sigma_inv = get_inverse_2x2(Covariance_Sigma)

    print(f"Determinant of Sigma: {det_Sigma}")
    print(f"Inverse of Sigma: {Sigma_inv}")

    # Manual calculation of y.T @ Sigma_inv @ y
    temp_vec = multiply_matrix_vector_2x2(Sigma_inv, y_test_point)
    quadratic_form_y = dot_product_vectors(y_test_point, temp_vec)

    # Calculate PDF using the general Gaussian formula
    gaussian_coeff = 1.0 / (2 * math.pi * math.sqrt(det_Sigma))
    gaussian_exponent = -0.5 * quadratic_form_y
    pdf_y_gaussian_formula = gaussian_coeff * math.exp(gaussian_exponent)
    
    print(f"f_Y({y_test_point}) using general Gaussian formula = {pdf_y_gaussian_formula:.6f}")

    tolerance = 1e-6
    if abs(pdf_y_val - pdf_y_gaussian_formula) < tolerance:
        print("\nVerification successful: Results from change-of-variables match general Gaussian formula.")
    else:
        print("\nVerification MISMATCH: Results differ. Check calculations.")

except ValueError as e:
    print(f"Error during calculations: {e}. Ensure matrix A is invertible.")
```
