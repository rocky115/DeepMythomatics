## Constrained Optimization and Lagrange Multipliers

In the previous section, we considered the problem of solving for the minimum of a function

$$
\min_x f(x), \tag{7.16}
$$

where $f: \mathbb{R}^D \to \mathbb{R}$. In this section, we have additional constraints. That is, for real-valued functions $g_i: \mathbb{R}^D \to \mathbb{R}$ for $i = 1, \ldots, m$, we consider the constrained optimization problem (see Figure 7.4 for an illustration)

$$
\min_x f(x) \tag{7.17}
$$

subject to $g_i(x) \leq 0$ for all $i = 1, \ldots, m$.

It is worth pointing out that the functions $f$ and $g_i$ could be non-convex in general, and we will consider the convex case in the next section.

One obvious, but not very practical, way of converting the constrained problem (7.17) into an unconstrained one is to use an indicator function

$$
J(x) = f(x) + \sum_{i=1}^m 1(g_i(x)), \tag{7.18}
$$

where $1(z)$ is an infinite step function

$$
1(z) = \begin{cases} 
0 & \text{if } z \leq 0 \\
\infty & \text{otherwise}
\end{cases}. \tag{7.19}
$$

This gives infinite penalty if the constraint is not satisfied, and hence would provide the same solution. However, this infinite step function is equally difficult to optimize.

We can overcome this difficulty by introducing Lagrange multipliers. The idea of Lagrange multipliers is to replace the step function with a linear function. We associate to problem (7.17) the Lagrangian by introducing the Lagrange multipliers $\lambda_i \geq 0$ corresponding to each inequality constraint respectively (Boyd and Vandenberghe, 2004, chapter 4) so that

$$
L(x, \lambda) = f(x) + \sum_{i=1}^m \lambda_i g_i(x) \tag{7.20a}
$$

$$
= f(x) + \lambda^\top g(x), \tag{7.20b}
$$

where in the last line we have concatenated all constraints $g_i(x)$ into a vector $g(x)$, and all the Lagrange multipliers into a vector $\lambda \in \mathbb{R}^m$.

We now introduce the idea of Lagrangian duality. In general, duality in optimization is the idea of converting an optimization problem in one set of variables $x$ (called the *primal variables*), into another optimization problem in a different set of variables $\lambda$ (called the *dual variables*). We introduce two different approaches to duality: In this section, we discuss Lagrangian duality; in Section 7.3.3, we discuss Legendre-Fenchel duality.

## Definition 7.1

The problem in (7.17)

$$
\min_x f(x) \tag{7.21}
$$

subject to $g_i(x) \leq 0$ for all $i = 1, \ldots, m$

is known as the *primal problem*, corresponding to the primal variables $x$. The associated *Lagrangian dual problem* is given by

$$
\max_{\lambda \in \mathbb{R}^m} D(\lambda) \tag{7.22}
$$

subject to $\lambda \geq 0$,

where $\lambda$ are the dual variables and $D(\lambda) = \min_{x \in \mathbb{R}^d} L(x, \lambda)$.

## Remark

In the discussion of Definition 7.1, we use two concepts that are also of independent interest (Boyd and Vandenberghe, 2004). First is the minimax inequality, which says that for any function with two arguments $\phi(x, y)$, the maximin is less than the minimax, i.e.,

$$
\max_y \min_x \phi(x, y) \leq \min_x \max_y \phi(x, y). \tag{7.23}
$$

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Objective Function and Constraint ---
def f(x1, x2):
    """Objective function: f(x1, x2) = x1^2 + x2^2."""
    return x1**2 + x2**2

def g(x1, x2):
    """Constraint function: g(x1, x2) = x1 + x2 - 1 <= 0."""
    return x1 + x2 - 1

# --- Lagrangian and Gradients ---
def lagrangian(x1, x2, lambda_):
    """Lagrangian: L(x1, x2, lambda) = f(x1, x2) + lambda * g(x1, x2)."""
    return f(x1, x2) + lambda_ * g(x1, x2)

def gradient_lagrangian(x1, x2, lambda_):
    """Compute the gradient of the Lagrangian w.r.t. x1, x2, and lambda."""
    # L/x1 = 2x1 + lambda
    grad_x1 = 2 * x1 + lambda_
    # L/x2 = 2x2 + lambda
    grad_x2 = 2 * x2 + lambda_
    # L/lambda = x1 + x2 - 1
    grad_lambda = x1 + x2 - 1
    return grad_x1, grad_x2, grad_lambda

# --- Solve the Constrained Optimization Problem ---
def solve_constrained_optimization():
    """
    Solve the constrained optimization problem using Lagrange multipliers.
    min x1^2 + x2^2 subject to x1 + x2 - 1 <= 0.
    """
    print("=== Constrained Optimization with Lagrange Multipliers ===")
    print("Section 7.2: Solving min x1^2 + x2^2 subject to x1 + x2 - 1 <= 0\n")

    # Case 1: Constraint is active (g(x) = 0, i.e., x1 + x2 = 1)
    print("Case 1: Constraint active (x1 + x2 = 1)")
    # Solve the system of equations:
    # 2x1 + lambda = 0
    # 2x2 + lambda = 0
    # x1 + x2 - 1 = 0
    # From the first two equations: 2x1 = 2x2 => x1 = x2
    # From the third equation: x1 + x2 = 1 => 2x1 = 1 => x1 = 0.5, x2 = 0.5
    # Substitute into the first equation: 2(0.5) + lambda = 0 => lambda = -1
    x1_active = 0.5
    x2_active = 0.5
    lambda_active = -1
    print(f"Solution: x1 = {x1_active}, x2 = {x2_active}, lambda = {lambda_active}")
    # Check if lambda >= 0 (KKT condition)
    if lambda_active >= 0:
        print("lambda >= 0, KKT conditions satisfied.")
        value_active = f(x1_active, x2_active)
        print(f"Objective value: {value_active:.4f}")
    else:
        print("lambda < 0, KKT conditions not satisfied. Constraint may not be active.")
    print()

    # Case 2: Constraint is inactive (g(x) < 0, i.e., lambda = 0)
    print("Case 2: Constraint inactive (lambda = 0)")
    # If lambda = 0, then:
    # 2x1 = 0 => x1 = 0
    # 2x2 = 0 => x2 = 0
    # Check constraint: x1 + x2 - 1 = 0 + 0 - 1 = -1 < 0
    x1_inactive = 0
    x2_inactive = 0
    lambda_inactive = 0
    constraint_value = g(x1_inactive, x2_inactive)
    print(f"Solution: x1 = {x1_inactive}, x2 = {x2_inactive}, lambda = {lambda_inactive}")
    print(f"Constraint value: {constraint_value:.4f} (should be <= 0)")
    if constraint_value <= 0:
        print("Constraint satisfied.")
        value_inactive = f(x1_inactive, x2_inactive)
        print(f"Objective value: {value_inactive:.4f}")
    else:
        print("Constraint not satisfied.")
    print()

    # Compare solutions
    if lambda_active >= 0 and g(x1_active, x2_active) == 0:
        optimal_x1, optimal_x2 = x1_active, x2_active
        optimal_value = value_active
    elif constraint_value <= 0:
        optimal_x1, optimal_x2 = x1_inactive, x2_inactive
        optimal_value = value_inactive
    else:
        optimal_x1, optimal_x2 = None, None
        optimal_value = None

    print("Optimal Solution:")
    if optimal_x1 is not None:
        print(f"x1 = {optimal_x1:.4f}, x2 = {optimal_x2:.4f}")
        print(f"Objective value: {optimal_value:.4f}")
    else:
        print("No feasible solution found.")
    print()

    return optimal_x1, optimal_x2

# --- Plotting Function ---
def plot_optimization():
    """Plot the objective function contours, constraint, and optimal solution."""
    optimal_x1, optimal_x2 = solve_constrained_optimization()

    # Generate grid for contour plot
    x1 = [i / 50.0 - 0.5 for i in range(75)]  # Range: -0.5 to 1.0
    x2 = [i / 50.0 - 0.5 for i in range(75)]
    z = [[f(x1i, x2j) for x2j in x2] for x1i in x1]

    plt.figure(figsize=(8, 6))
    # Plot contours of the objective function
    plt.contour(x1, x2, z, levels=20, cmap="viridis")
    plt.colorbar(label="f(x1, x2) = x1^2 + x2^2")

    # Plot the constraint boundary x1 + x2 = 1
    x1_constraint = [i / 50.0 - 0.5 for i in range(75)]
    x2_constraint = [1 - xi for xi in x1_constraint]
    plt.plot(x1_constraint, x2_constraint, "r--", label="Constraint: x1 + x2 = 1")

    # Plot the feasible region (shade below the constraint)
    plt.fill_between(x1_constraint, x2_constraint, -0.5, color="gray", alpha=0.3, label="Feasible region (x1 + x2 <= 1)")

    # Plot the optimal solution
    if optimal_x1 is not None and optimal_x2 is not None:
        plt.scatter([optimal_x1], [optimal_x2], color="red", s=100, label="Optimal solution")

    plt.title("Constrained Optimization: min x1^2 + x2^2 subject to x1 + x2 <= 1")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Constrained Optimization with Lagrange Multipliers and Seaborn Plotting")
    print("=" * 60)

    # Plot the optimization problem
    print("Plotting Optimization Problem")
    plot_optimization()

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(" Solved constrained optimization problem using Lagrange multipliers (Section 7.2)")
    print(" Plotted objective function contours, constraint, and optimal solution")
```

This inequality can be proved by considering the inequality

$$
\forall x, y \quad \min_x \phi(x, y) \leq \max_y \phi(x, y). \tag{7.24}
$$

Note that taking the maximum over $y$ of the left-hand side of (7.24) maintains the inequality since the inequality is true for all $y$. Similarly, we can take the minimum over $x$ of the right-hand side of (7.24) to obtain (7.23).

The second concept is *weak duality*, which uses (7.23) to show that primal values are always greater than or equal to dual values. This is described in more detail in (7.27). $\diamond$

Recall that the difference between $J(x)$ in (7.18) and the Lagrangian in (7.20b) is that we have relaxed the indicator function to a linear function. Therefore, when $\lambda \geq 0$, the Lagrangian $L(x, \lambda)$ is a lower bound of $J(x)$. Hence, the maximum of $L(x, \lambda)$ with respect to $\lambda$ is

$$
J(x) = \max_{\lambda \geq 0} L(x, \lambda). \tag{7.25}
$$

Recall that the original problem was minimizing $J(x)$,

$$
\min_{x \in \mathbb{R}^d} \max_{\lambda \geq 0} L(x, \lambda). \tag{7.26}
$$

By the minimax inequality (7.23), it follows that swapping the order of the minimum and maximum results in a smaller value, i.e.,

$$
\min_{x \in \mathbb{R}^d} \max_{\lambda \geq 0} L(x, \lambda) \geq \max_{\lambda \geq 0} \min_{x \in \mathbb{R}^d} L(x, \lambda). \tag{7.27}
$$

This is also known as *weak duality*. Note that the inner part of the right-hand side is the dual objective function $D(\lambda)$ and the definition follows. In contrast to the original optimization problem, which has constraints, $\min_{x \in \mathbb{R}^d} L(x, \lambda)$ is an unconstrained optimization problem for a given value of $\lambda$. If solving $\min_{x \in \mathbb{R}^d} L(x, \lambda)$ is easy, then the overall problem is easy to solve. We can see this by observing from (7.20b) that $L(x, \lambda)$ is affine with respect to $\lambda$. Therefore $\min_{x \in \mathbb{R}^d} L(x, \lambda)$ is a pointwise minimum of affine functions of $\lambda$, and hence $D(\lambda)$ is concave even though $f(\cdot)$ and $g_i(\cdot)$ may be nonconvex. The outer problem, maximization over $\lambda$, is the maximum of a concave function and can be efficiently computed.

Assuming $f(\cdot)$ and $g_i(\cdot)$ are differentiable, we find the Lagrange dual problem by differentiating the Lagrangian with respect to $x$, setting the differential to zero, and solving for the optimal value. We will discuss two concrete examples in Sections 7.3.1 and 7.3.2, where $f(\cdot)$ and $g_i(\cdot)$ are convex.

## Remark (Equality Constraints)

Consider (7.17) with additional equality constraints

$$
\min_x f(x)
$$

subject to

$$
g_i(x) \leq 0 \quad \text{for all } i = 1, \ldots, m \tag{7.28}
$$

$$
h_j(x) = 0 \quad \text{for all } j = 1, \ldots, n.
$$

We can model equality constraints by replacing them with two inequality constraints. That is, for each equality constraint $h_j(x) = 0$, we equivalently replace it by two constraints $h_j(x) \leq 0$ and $h_j(x) \geq 0$. It turns out that the resulting Lagrange multipliers are then unconstrained. Therefore, we constrain the Lagrange multipliers corresponding to the inequality constraints in (7.28) to be non-negative, and leave the Lagrange multipliers corresponding to the equality constraints unconstrained. $\diamond$

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Objective Function and Constraints ---
def f(x1, x2):
    """Objective function: f(x1, x2) = x1^2 + x2^2."""
    return x1**2 + x2**2

def g(x1, x2):
    """Inequality constraint: g(x1, x2) = x1 + x2 - 2 <= 0."""
    return x1 + x2 - 2

def h(x1, x2):
    """Equality constraint: h(x1, x2) = x1 - x2 = 0."""
    return x1 - x2

# --- Lagrangian ---
def lagrangian(x1, x2, lambda_, mu):
    """Lagrangian: L(x1, x2, lambda, mu) = f + lambda * g + mu * h."""
    return f(x1, x2) + lambda_ * g(x1, x2) + mu * h(x1, x2)

# --- Solve the Primal Problem ---
def solve_primal():
    """
    Solve the primal problem: min x1^2 + x2^2
    subject to x1 + x2 - 2 <= 0, x1 - x2 = 0.
    """
    print("=== Primal Problem Solution ===")
    # Gradients of Lagrangian:
    # L/x1 = 2x1 + lambda + mu = 0  (1)
    # L/x2 = 2x2 + lambda - mu = 0  (2)
    # L/lambda = x1 + x2 - 2 = 0    (3) (active constraint)
    # L/mu = x1 - x2 = 0            (4)
    # From (4): x1 = x2
    # Substitute into (3): 2x1 = 2 => x1 = 1, x2 = 1
    # Substitute into (1) and (2):
    # 2(1) + lambda + mu = 0 => 2 + lambda + mu = 0  (5)
    # 2(1) + lambda - mu = 0 => 2 + lambda - mu = 0  (6)
    # Add (5) and (6): 4 + 2lambda = 0 => lambda = -2
    # Since lambda < 0, constraint may not be active, set lambda = 0 and re-solve:
    # lambda = 0 => 2 + mu = 0 => mu = -2
    # But let's try inactive constraint (lambda = 0):
    # If lambda = 0:
    # 2x1 + mu = 0, 2x2 - mu = 0, x1 = x2
    # => 2x1 + mu = 0, 2x1 - mu = 0 => 4x1 = 0 => x1 = 0, x2 = 0, mu = 0
    # Check constraint: 0 + 0 - 2 = -2 <= 0 (satisfied)

    x1_opt = 0
    x2_opt = 0
    lambda_opt = 0
    mu_opt = 0
    primal_value = f(x1_opt, x2_opt)

    print(f"Optimal solution: x1 = {x1_opt:.4f}, x2 = {x2_opt:.4f}")
    print(f"Lambda (inequality): {lambda_opt:.4f}, Mu (equality): {mu_opt:.4f}")
    print(f"Primal objective value: {primal_value:.4f}")
    print(f"Inequality constraint value: {g(x1_opt, x2_opt):.4f} (should be <= 0)")
    print(f"Equality constraint value: {h(x1_opt, x2_opt):.4f} (should be = 0)")
    print()

    return x1_opt, x2_opt, primal_value

# --- Dual Function and Weak Duality ---
def dual_function(lambda_, mu):
    """
    Compute the dual function D(lambda, mu) = min_{x1, x2} L(x1, x2, lambda, mu).
    """
    # Minimize L w.r.t. x1, x2:
    # L/x1 = 2x1 + lambda + mu = 0 => 2x1 = -(lambda + mu) => x1 = -(lambda + mu)/2
    # L/x2 = 2x2 + lambda - mu = 0 => 2x2 = -(lambda - mu) => x2 = -(lambda - mu)/2
    x1 = -(lambda_ + mu) / 2
    x2 = -(lambda_ - mu) / 2
    # Substitute back into Lagrangian
    return lagrangian(x1, x2, lambda_, mu)

def solve_dual():
    """
    Solve the dual problem: max_{lambda >= 0, mu} D(lambda, mu).
    """
    print("=== Dual Problem Solution ===")
    # Since the constraint was inactive in the primal, try lambda = 0
    # D(lambda, mu) = min L
    # Let's find the maximum by trying some values (simple grid search for lambda, mu)
    best_lambda = 0
    best_mu = 0
    best_dual_value = float("-inf")

    # Since lambda >= 0, and mu is unconstrained, try a small range
    lambda_values = [0, 0.5, 1, 2]
    mu_values = [-2, -1, 0, 1, 2]

    for lambda_ in lambda_values:
        for mu in mu_values:
            if lambda_ < 0:
                continue
            dual_val = dual_function(lambda_, mu)
            if dual_val > best_dual_value:
                best_dual_value = dual_val
                best_lambda = lambda_
                best_mu = mu

    print(f"Optimal dual variables: lambda = {best_lambda:.4f}, mu = {best_mu:.4f}")
    print(f"Dual objective value: {best_dual_value:.4f}")
    print()

    return best_dual_value

# --- Plotting Function ---
def plot_optimization():
    """Plot the objective function contours, constraints, and optimal solution."""
    x1_opt, x2_opt, primal_value = solve_primal()
    dual_value = solve_dual()

    print("=== Weak Duality Verification ===")
    print(f"Primal value: {primal_value:.4f}")
    print(f"Dual value: {dual_value:.4f}")
    print(f"Primal >= Dual: {primal_value >= dual_value}\n")

    # Generate grid for contour plot
    x1 = [i / 50.0 - 0.5 for i in range(125)]  # Range: -0.5 to 2.0
    x2 = [i / 50.0 - 0.5 for i in range(125)]
    z = [[f(x1i, x2j) for x2j in x2] for x1i in x1]

    plt.figure(figsize=(8, 6))
    # Plot contours of the objective function
    plt.contour(x1, x2, z, levels=20, cmap="viridis")
    plt.colorbar(label="f(x1, x2) = x1^2 + x2^2")

    # Plot the inequality constraint boundary x1 + x2 = 2
    x1_ineq = [i / 50.0 - 0.5 for i in range(125)]
    x2_ineq = [2 - xi for xi in x1_ineq]
    plt.plot(x1_ineq, x2_ineq, "r--", label="Inequality: x1 + x2 = 2")

    # Plot the equality constraint x1 = x2
    x1_eq = [i / 50.0 - 0.5 for i in range(125)]
    x2_eq = x1_eq
    plt.plot(x1_eq, x2_eq, "b--", label="Equality: x1 = x2")

    # Plot the feasible region (shade below the inequality constraint and on the equality line)
    plt.fill_between(x1_ineq, x2_ineq, -0.5, where=[xi <= x2i for xi, x2i in zip(x1_ineq, x2_ineq)], color="gray", alpha=0.3)

    # Plot the optimal solution
    plt.scatter([x1_opt], [x2_opt], color="red", s=100, label="Optimal solution")

    plt.title("Constrained Optimization: min x1^2 + x2^2\nsubject to x1 + x2 <= 2, x1 = x2")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Constrained Optimization with Lagrange Multipliers and Seaborn Plotting")
    print("=" * 60)

    # Plot the optimization problem
    print("Plotting Optimization Problem")
    plot_optimization()

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(" Solved constrained optimization with inequality and equality constraints (Section 7.2)")
    print(" Verified weak duality (Equation 7.27)")
    print(" Plotted objective function contours, constraints, and optimal solution")
```

# 7.3 Convex Optimization

We focus our attention on a particularly useful class of optimization problems, where we can guarantee global optimality. When $f(\cdot)$ is a convex function, and when the constraints involving $g(\cdot)$ and $h(\cdot)$ define convex sets, this is called a *convex optimization problem*. In this setting, we have *strong duality*: the optimal solution of the dual problem is the same as the optimal solution of the primal problem. The distinction between convex functions and convex sets are often not strictly stated in machine learning literature, but one can often infer the implied meaning from context.

## Definition 7.2

A set $C$ is a convex set if for any $x, y \in C$ and for any scalar $\theta$ with $0 \leq \theta \leq 1$, we have

$$
\theta x + (1 - \theta) y \in C. \tag{7.29}
$$

Convex sets are sets such that a straight line connecting any two elements of the set lies inside the set. Figures 7.5 and 7.6 illustrate convex and non-convex sets, respectively.

Convex functions are functions such that a straight line between any two points of the function lies above the function. Figure 7.2 shows a non-convex function, and Figure 7.3 shows a convex function. Another convex function is shown in Figure 7.7.

## Definition 7.3

Let function $f: \mathbb{R}^D \to \mathbb{R}$ be a function whose domain is a convex set. The function $f$ is a *convex function* if for all $x, y$ in the domain of $f\), and for any scalar \(\theta$ with \(0 \leq \theta \leq 1\),

$$
f(\theta x + (1 - \theta) y) \leq \theta f(x) + (1 - \theta) f(y). \tag{7.30}
$$


Fig.7 Example of a convex function.

## Remark

A concave function is the negative of a convex function. $\diamond$

The constraints involving $g(\cdot)$ and $h(\cdot)$ in (7.28) truncate functions at a scalar value, resulting in sets. Another relation between convex functions and convex sets is to consider the set obtained by filling in a convex function. A convex function is a bowl-like object, and we imagine pouring water into it to fill it up. This resulting filled-in set, called the *epigraph* of the convex function, is a convex set.

If a function $f: \mathbb{R}^n \to \mathbb{R}$ is differentiable, we can specify convexity in terms of its gradient $\nabla_x f(x)$ (Section 5.2). A function $f(x)$ is convex if and only if for any two points $x, y$ it holds that

$$
f(y) \geq f(x) + \nabla_x f(x)^\top (y - x). \tag{7.31}
$$

If we further know that a function $f$ is twice differentiable, that is, the Hessian (5.147) exists for all values in the domain of $x$, then the function $f(x)$ is convex if and only if $\nabla_x^2 f(x)$ is positive semidefinite (Boyd and Vandenberghe, 2004).

## Example 7.3

The negative entropy function $f(x) = x \log_2 x$ is convex for $x > 0$. A visualization of the function is shown in Figure 7.8, and we can see that the function is convex.

To illustrate the previous definitions of convexity, let us check the calculations for two points $x = 2$ and $x = 4$. Note that to prove convexity of $f(x)$ we would need to check for all points $x \in \mathbb{R}$. Recall Definition 7.3. Consider a point midway between the two points (that is $\theta = 0.5$); then the left-hand side is

$$
f(0.5 \cdot 2 + 0.5 \cdot 4) = 3 \log_2 3 \approx 4.75.
$$

The right-hand side is

$$
0.5 (2 \log_2 2) + 0.5 (4 \log_2 4) = 1 + 4 = 5.
$$

And therefore the definition is satisfied.

Since $f(x)$ is differentiable, we can alternatively use (7.31). Calculating the derivative of $f(x)$, we obtain

$$
\nabla_x (x \log_2 x) = 1 \cdot \log_2 x + x \cdot \frac{1}{x \log_e 2} = \log_2 x + \frac{1}{\log_e 2}. \tag{7.32}
$$

Using the same two test points $x = 2$ and $x = 4$, the left-hand side of (7.31) is given by $f(4) = 8$. The right-hand side is

$$
f(x) + \nabla_x^\top (y - x) = f(2) + \nabla f(2) \cdot (4 - 2) \tag{7.33a}
$$

$$
= 2 + \left(1 + \frac{1}{\log_e 2}\right) \cdot 2 \approx 6.9. \tag{7.33b}
$$

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better visuals
sns.set(style="whitegrid")

# --- Negative Entropy Function and Its Gradient ---
def negative_entropy(x):
    """Negative entropy function: f(x) = x log_2 x for x > 0 (Example 7.3)."""
    if x <= 0:
        return float("inf")  # Undefined for x <= 0
    return x * math.log2(x)

def gradient_negative_entropy(x):
    """Gradient of negative entropy: log_2 x + 1/ln(2) (Equation 7.32)."""
    if x <= 0:
        return float("inf")
    return math.log2(x) + 1 / math.log(math.e, 2)

# --- Check Convexity Using Definition 7.3 ---
def check_convexity_definition(f, x, y, theta=0.5):
    """
    Check convexity using Definition 7.3 (Equation 7.30):
    f(theta * x + (1 - theta) * y) <= theta * f(x) + (1 - theta) * f(y).
    """
    left = f(theta * x + (1 - theta) * y)
    right = theta * f(x) + (1 - theta) * f(y)
    return left <= right, left, right

# --- Check Convexity Using Gradient Condition ---
def check_convexity_gradient(f, grad_f, x, y):
    """
    Check convexity using gradient condition (Equation 7.31):
    f(y) >= f(x) + grad_f(x) * (y - x).
    """
    left = f(y)
    right = f(x) + grad_f(x) * (y - x)
    return left >= right, left, right

# --- Check Convex Set (Definition 7.2) ---
def is_convex_set(x, y, theta=0.5):
    """
    Check if the set {x | x >= 0} is convex using Definition 7.2 (Equation 7.29).
    For points x, y in the set, theta * x + (1 - theta) * y should be in the set.
    """
    if x < 0 or y < 0:
        return False, None
    interpolated_point = theta * x + (1 - theta) * y
    return interpolated_point >= 0, interpolated_point

# --- Demonstration ---
def demonstrate_convexity():
    """
    Demonstrate convexity of the negative entropy function and a convex set.
    - Verify convexity of f(x) = x log_2 x (Example 7.3)
    - Check if the set {x | x >= 0} is convex (Definition 7.2)
    """
    print("=== Convexity Analysis ===")
    print("Section 7.3: Convex Optimization and Convexity Verification\n")

    # Test points from Example 7.3
    x, y = 2, 4
    theta = 0.5

    # Check convexity using Definition 7.3
    print("Verifying Convexity of f(x) = x log_2 x Using Definition 7.3:")
    is_convex, left, right = check_convexity_definition(negative_entropy, x, y, theta)
    print(f"x = {x}, y = {y}, theta = {theta}")
    print(f"f({theta * x + (1 - theta) * y}) = {left:.4f}")
    print(f"{theta} * f({x}) + (1 - {theta}) * f({y}) = {right:.4f}")
    print(f"Is convex (f(theta * x + (1 - theta) * y) <= theta * f(x) + (1 - theta) * f(y))? {is_convex}\n")

    # Check convexity using gradient condition
    print("Verifying Convexity Using Gradient Condition (Equation 7.31):")
    is_convex_grad, left_grad, right_grad = check_convexity_gradient(negative_entropy, gradient_negative_entropy, x, y)
    print(f"x = {x}, y = {y}")
    print(f"f({y}) = {left_grad:.4f}")
    print(f"f({x}) + grad_f({x}) * ({y} - {x}) = {right_grad:.4f}")
    print(f"Is convex (f(y) >= f(x) + grad_f(x) * (y - x))? {is_convex_grad}\n")

    # Check if the set {x | x >= 0} is convex
    print("Verifying Convex Set {x | x >= 0} Using Definition 7.2:")
    x_set, y_set = 2, 4
    is_convex_set_result, interpolated_point = is_convex_set(x_set, y_set, theta)
    print(f"x = {x_set}, y = {y_set}, theta = {theta}")
    print(f"Interpolated point: {theta} * {x_set} + (1 - {theta}) * {y_set} = {interpolated_point:.4f}")
    print(f"Is in set (interpolated point >= 0)? {is_convex_set_result}")

# --- Plotting Function ---
def plot_negative_entropy():
    """Plot the negative entropy function and illustrate convexity."""
    x_values = [i / 100 for i in range(1, 501)]  # x from 0.01 to 5
    y_values = [negative_entropy(x) for x in x_values]

    # Points for convexity illustration
    x1, x2 = 2, 4
    y1, y2 = negative_entropy(x1), negative_entropy(x2)
    theta = 0.5
    x_mid = theta * x1 + (1 - theta) * x2
    y_mid = negative_entropy(x_mid)
    y_interp = theta * y1 + (1 - theta) * y2

    plt.figure(figsize=(8, 6))
    sns.lineplot(x=x_values, y=y_values, label="f(x) = x log_2 x", color="teal")

    # Plot the line segment between (x1, f(x1)) and (x2, f(x2))
    plt.plot([x1, x2], [y1, y2], "r--", label=f"Line between ({x1}, {y1:.2f}) and ({x2}, {y2:.2f})")
    plt.scatter([x1, x2], [y1, y2], color="red", s=50)
    plt.scatter([x_mid], [y_mid], color="purple", s=50, label=f"f({x_mid}) = {y_mid:.2f}")
    plt.scatter([x_mid], [y_interp], color="orange", s=50, label=f"Interpolated: {y_interp:.2f}")

    plt.title("Negative Entropy Function: Illustrating Convexity")
    plt.xlabel("x")
    plt.ylabel("f(x) = x log_2 x")
    plt.legend()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("Convexity Analysis with Seaborn Plotting")
    print("=" * 60)

    # Run demonstration
    demonstrate_convexity()

    # Plot the negative entropy function
    print("\nPlotting Negative Entropy Function")
    plot_negative_entropy()

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(" Verified convexity of f(x) = x log_2 x using Definition 7.3 (Equation 7.30)")
    print(" Verified convexity using gradient condition (Equation 7.31)")
    print(" Confirmed that the set {x | x >= 0} is convex (Definition 7.2)")
    print(" Plotted f(x) = x log_2 x with illustration of convexity")
```

## Example 4

A nonnegative weighted sum of convex functions is convex. Observe that if $f$ is a convex function, and $\alpha \geq 0$ is a nonnegative scalar, then the function $\alpha f$ is convex. We can see this by multiplying $\alpha$ to both sides of the equation in Definition 7.3, and recalling that multiplying a nonnegative number does not change the inequality.

If $f_1$ and $f_2$ are convex functions, then we have by the definition

$$
f_1(\theta x + (1 - \theta) y) \leq \theta f_1(x) + (1 - \theta) f_1(y) \tag{7.34}
$$

$$
f_2(\theta x + (1 - \theta) y) \leq \theta f_2(x) + (1 - \theta) f_2(y). \tag{7.35}
$$

Summing up both sides gives us

$$
f_1(\theta x + (1 - \theta) y) + f_2(\theta x + (1 - \theta) y) \leq \theta f_1(x) + (1 - \theta) f_1(y) + \theta f_2(x) + (1 - \theta) f_2(y), \tag{7.36}
$$

where the right-hand side can be rearranged to

$$
\theta (f_1(x) + f_2(x)) + (1 - \theta) (f_1(y) + f_2(y)), \tag{7.37}
$$

completing the proof that the sum of convex functions is convex. Combining the preceding two facts, we see that $\alpha f_1(x) + \beta f_2(x)$ is convex for $\alpha, \beta \geq 0$. This closure property can be extended using a similar argument for nonnegative weighted sums of more than two convex functions.

## Remark

The inequality in (7.30) is sometimes called *Jensens inequality*. In fact, a whole class of inequalities for taking nonnegative weighted sums of convex functions are all called Jensens inequality. $\diamond$

In summary, a constrained optimization problem is called a *convex optimization problem* if

$$
\min_x f(x)
$$

subject to

$$
g_i(x) \leq 0 \quad \text{for all } i = 1, \ldots, m \tag{7.38}
$$

$$
h_j(x) = 0 \quad \text{for all } j = 1, \ldots, n,
$$

where all functions $f(x)$ and $g_i(x)$ are convex functions, and all $h_j(x) = 0$ define convex sets.

In the following, we will describe two classes of convex optimization problems that are widely used and well understood.

## Tangent Line to \(f(x) = x \log_2 x\) at \(x = 2\)

The function \(f(x) = x \log_2 x\) (negative entropy, Example 7.3) is plotted over the interval \([0, 5]\), as shown in the figure context (with \(y\)-axis ranging from 0 to 10). We compute the tangent line to \(f(x)\) at \(x = 2\).

The gradient of \(f(x)\) at \(x = 2\) is given by (Equation 7.32):

$$
\nabla_x (x \log_2 x) = \log_2 x + \frac{1}{\log_e 2}.
$$

At \(x = 2\):

$$
\nabla_x f(2) = \log_2 2 + \frac{1}{\log_e 2} = 1 + \frac{1}{\log_e 2} \approx 2.4427.
$$

The function value at \(x = 2\) is:

$$
f(2) = 2 \log_2 2 = 2 \cdot 1 = 2.
$$

The equation of the tangent line at \(x = 2\) is:

$$
y = f(2) + \nabla_x f(2) (x - 2) = 2 + \left(1 + \frac{1}{\log_e 2}\right) (x - 2).
$$

This tangent line touches the curve at \((2, 2)\) and has a slope of approximately 2.4427, illustrating the convexity of \(f(x)\) since the function lies above the tangent line (Equation 7.31).
