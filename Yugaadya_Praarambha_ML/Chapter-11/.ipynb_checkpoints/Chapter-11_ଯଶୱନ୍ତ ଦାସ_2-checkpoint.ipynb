##  Traditional Derivation of the Margin

In Section 12.2.1, we assumed $\|w\| = 1$ to focus on direction only. Here, we take a different approach: we scale the data so that the closest example $x_a$ satisfies:

$$
\langle w, x_a \rangle + b = 1
$$

Since $x_a'$ is the orthogonal projection of $x_a$ onto the hyperplane, by definition:

$$
\langle w, x_a' \rangle + b = 0 \tag{12.11}
$$

Substituting the vector decomposition $x_a = x_a' + \frac{r}{\|w\|}w$ into the equation:

$$
\left\langle w, x_a - \frac{r}{\|w\|}w \right\rangle + b = 0 \tag{12.12}
$$

By bilinearity of the inner product:

$$
\langle w, x_a \rangle + b - r \|w\| = 0 \tag{12.13}
$$

Since $\langle w, x_a \rangle + b = 1$ by assumption, we solve for $r$:

$$
r = \frac{1}{\|w\|} \tag{12.14}
$$

---

## Margin Constraints

We require that **all** examples be at least distance $1$ away from the hyperplane on the correct side. This yields:

$$
y_n(\langle w, x_n \rangle + b) \geq 1 \tag{12.15}
$$

---

## Optimization Formulation

Maximizing the margin is equivalent to minimizing the norm of $w$:

$$
\begin{aligned}
\min_{w,b} \quad & \frac{1}{\|w\|} \tag{12.16} \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1 \quad \forall n \tag{12.17}
\end{aligned}
$$

Instead of maximizing the reciprocal, we minimize the squared norm (which is convex):

$$
\begin{aligned}
\min_{w,b} \quad & \frac{1}{2} \|w\|^2 \tag{12.18} \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1 \quad \forall n \tag{12.19}
\end{aligned}
$$

This is the standard formulation of the **hard margin SVM**, which does not allow for margin violations.


````python
import random
import matplotlib.pyplot as plt

# Generate linearly separable dataset
def generate_data(n=50):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        label = 1 if x1 + x2 > 0.5 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product
def dot(w, x):
    return sum(wi * xi for wi, xi in zip(w, x))

# Gradient descent for hard-margin SVM
def train_svm(X, y, lr=0.01, epochs=1000, lambda_reg=0.01):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < 1:
                w = [wi - lr * (2 * lambda_reg * wi - yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * yi
            else:
                w = [wi - lr * (2 * lambda_reg * wi) for wi in w]
    return w, b

# Visualize decision boundary and margin
def plot_svm(X, y, w, b):
    import math
    def norm(v): return math.sqrt(dot(v, v))

    plt.figure(figsize=(6, 6))
    for xi, yi in zip(X, y):
        plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')

    x_vals = [-2.5, 2.5]
    y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
    margin = 1 / norm(w)
    y_margin_up = [y + margin for y in y_vals]
    y_margin_down = [y - margin for y in y_vals]

    plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
    plt.plot(x_vals, y_margin_up, 'r--', alpha=0.6, label='Margin')
    plt.plot(x_vals, y_margin_down, 'r--', alpha=0.6)
    plt.title("Hard-Margin SVM: Traditional Margin Scaling")
    plt.xlabel("x")
    plt.ylabel("x")
    plt.axis("equal")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run it all
X, y = generate_data()
w, b = train_svm(X, y)
plot_svm(X, y, w, b)
````

# 12.2.3 Why We Can Set the Margin to 1

In Section 12.2.1, we proposed maximizing a margin value $r$, representing the distance of the closest example to the hyperplane under the constraint $\|w\| = 1$.

In Section 12.2.2, we instead scaled the data so that the margin becomes $1$ for the closest point.

This section proves the two formulations are equivalent.

---

## Theorem 12.1

Maximizing the margin $r$ with normalized weights:

$$
\begin{aligned}
\max_{w, b, r} \quad & r \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq r, \quad \|w\| = 1, \quad r > 0
\end{aligned} \tag{12.20}
$$

is equivalent to minimizing the squared norm of $w$ with unit margin constraint:

$$
\begin{aligned}
\min_{w, b} \quad & \frac{1}{2} \|w\|^2 \\
\text{subject to} \quad & y_n(\langle w, x_n \rangle + b) \geq 1
\end{aligned} \tag{12.21}
$$

---


Fig.6 (a) Linearly separable and (b) non-linearly separable data.

## Proof Sketch

Start with the original problem (12.20). Since squaring is monotonic for non-negative inputs, we can equivalently maximize $r^2$:

$$
\max_{w, b, r} \quad r^2 \quad \text{subject to} \quad y_n(\langle w, x_n \rangle + b) \geq r, \; \|w\| = 1, \; r > 0
$$

Now, reparameterize the weight vector $w'$ without the normalization:

$$
\max_{w', b, r} \quad r^2 \quad \text{subject to} \quad y_n \left( \frac{\langle w', x_n \rangle + b}{\|w'\|} \right) \geq r, \; r > 0 \tag{12.22}
$$

Divide both sides of the inequality by $r > 0$:

$$
y_n \left( \frac{\langle w', x_n \rangle + b}{r \|w'\|} \right) \geq 1 \tag{12.23}
$$

Now define:

$$
w'' = \frac{w'}{r \|w'\|}, \quad b'' = \frac{b}{r \|w'\|} \tag{12.24}
$$

Substituting into the inequality:

$$
y_n(\langle w'', x_n \rangle + b'') \geq 1
$$

which matches the constraint in (12.21). Since

$$
\|w''\| = \frac{1}{r}
$$

maximizing $r$ is equivalent to minimizing $\|w''\|$, or more commonly:

$$
\min_{w'', b''} \quad \frac{1}{2} \|w''\|^2 \tag{12.25}
$$

which concludes the proof.


````python
import random
import matplotlib.pyplot as plt
import math

# Generate separable 2D data
def generate_data(n=50):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        label = 1 if x1 + x2 > 0.5 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product
def dot(w, x): return sum(wi * xi for wi, xi in zip(w, x))

# Norm of vector
def norm(w): return math.sqrt(dot(w, w))

# Unit norm version: maximize r s.t. ||w|| = 1
def train_fixed_norm(X, y, lr=0.1, epochs=200):
    w = [1.0, 1.0]
    b = 0.0
    r = 0.1
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < r:
                b += lr * yi
                r -= lr * 0.01
        w = [wi / norm(w) for wi in w]  # normalize w
    return w, b

# Standard SVM: minimize 1/2||w|| s.t. margin = 1
def train_unit_margin(X, y, lr=0.01, epochs=1000, lam=0.01):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            if yi * (dot(w, xi) + b) < 1:
                w = [wi - lr * (2 * lam * wi - yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * yi
            else:
                w = [wi - lr * 2 * lam * wi for wi in w]
    return w, b

# Plotting helper
def plot_models(X, y, models, titles):
    plt.figure(figsize=(10, 5))
    for idx, (w, b) in enumerate(models):
        plt.subplot(1, 2, idx+1)
        for xi, yi in zip(X, y):
            plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')
        x_vals = [-2.5, 2.5]
        y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
        plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
        plt.title(titles[idx])
        plt.axis("equal")
        plt.grid(True)
        plt.legend()
    plt.tight_layout()
    plt.show()

# Run both methods
X, y = generate_data()
w1, b1 = train_fixed_norm(X, y)
w2, b2 = train_unit_margin(X, y)

plot_models(X, y, [(w1, b1), (w2, b2)], ["Formulation: max margin (unit norm)", "Formulation: min norm (unit margin)"])
````

## Soft Margin SVM: Geometric View

In cases where the data is **not linearly separable**, we allow some examples to lie **within the margin** or even on the **wrong side** of the hyperplane. This approach leads to the **soft margin SVM**.

---

## Geometric Intuition

For each example-label pair $(x_n, y_n)$, we introduce a **slack variable** $\xi_n \geq 0$, which allows a point to:

- Fall within the margin, or
- Be misclassified (on the wrong side of the hyperplane).

We define the decision boundary as:

$$
f(x) = \langle w, x \rangle + b
$$

and adjust the margin constraint from:

$$
y_n (\langle w, x_n \rangle + b) \geq 1
$$

to:

$$
y_n (\langle w, x_n \rangle + b) \geq 1 - \xi_n \tag{12.26b}
$$

To keep $\xi_n$ meaningful, we constrain:

$$
\xi_n \geq 0 \tag{12.26c}
$$

---

## Soft Margin SVM Optimization Problem

We now minimize a combination of:

- **Margin penalty**: $\|w\|^2$
- **Total slack**: $\sum \xi_n$

The resulting **soft margin SVM** optimization problem is:

$$
\min_{w, b, \{\xi_n\}} \quad \frac{1}{2} \|w\|^2 + C \sum_{n=1}^N \xi_n \tag{12.26a}
$$

subject to:

$$
y_n (\langle w, x_n \rangle + b) \geq 1 - \xi_n \quad \text{for } n = 1, \dots, N \tag{12.26b}
$$

$$
\xi_n \geq 0 \quad \text{for } n = 1, \dots, N \tag{12.26c}
$$

---

### Remarks

- The hyperparameter $C > 0$ controls the **tradeoff** between margin size and classification accuracy on training data.
- A **large $C$** places more emphasis on minimizing classification errors (less regularization).
- The term $\|w\|^2$ is often called the **regularizer**, and $C$ is the **regularization parameter**.
- Note: In this formulation, $b$ is *not* regularized, which simplifies optimization but complicates theoretical analysis.

This formulation is sometimes called the **C-SVM**, especially in libraries like LIBSVM and sklearn.

---

In Section 12.2.5, well reformulate the soft margin SVM from the **loss function perspective**, connecting it to empirical risk minimization.


````python
import random
import matplotlib.pyplot as plt

# Generate synthetic 2D data (some overlap)
def generate_data(n=100):
    X, y = [], []
    for _ in range(n):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        noise = random.gauss(0, 0.5)
        label = 1 if x1 + x2 + noise > 0 else -1
        X.append([x1, x2])
        y.append(label)
    return X, y

# Dot product and norm helpers
def dot(w, x): return sum(wi * xi for wi, xi in zip(w, x))
def norm(w): return sum(wi ** 2 for wi in w) ** 0.5

# Train Soft Margin SVM with gradient descent (hinge loss + L2 regularization)
def train_soft_svm(X, y, lr=0.01, epochs=1000, C=1.0):
    w = [0.0, 0.0]
    b = 0.0
    for _ in range(epochs):
        for xi, yi in zip(X, y):
            margin = yi * (dot(w, xi) + b)
            if margin < 1:
                w = [wi - lr * (2 * wi - C * yi * xi[j]) for j, wi in enumerate(w)]
                b += lr * C * yi
            else:
                w = [wi - lr * 2 * wi for wi in w]
    return w, b

# Plot data, hyperplane, and margins
def plot_svm(X, y, w, b):
    plt.figure(figsize=(6, 6))
    for xi, yi in zip(X, y):
        plt.scatter(xi[0], xi[1], c='orange' if yi == 1 else 'blue', marker='x' if yi == 1 else 'o')

    # Decision boundary
    x_vals = [-2.5, 2.5]
    y_vals = [(-w[0] * x - b) / w[1] for x in x_vals]
    margin = 1 / (norm(w) + 1e-12)
    y_margin_up = [y + margin for y in y_vals]
    y_margin_down = [y - margin for y in y_vals]

    plt.plot(x_vals, y_vals, 'k--', label='Hyperplane')
    plt.plot(x_vals, y_margin_up, 'r--', alpha=0.6)
    plt.plot(x_vals, y_margin_down, 'r--', alpha=0.6)
    plt.title("Soft Margin SVM (Geometric View)")
    plt.xlabel("x")
    plt.ylabel("x")
    plt.axis("equal")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the pipeline
X, y = generate_data()
w, b = train_soft_svm(X, y, C=10.0)
plot_svm(X, y, w, b)
````

# Support Vector Machines: Loss Functions and Hinge Loss

## 1. Introduction to SVM Hypothesis Class

In Support Vector Machines, we choose **hyperplanes** as our hypothesis class:

$$f(x) = \langle w, x \rangle + b \quad (12.27)$$

where:
- $w$ is the weight vector
- $b$ is the bias term
- $\langle w, x \rangle$ denotes the inner product

The margin corresponds to the regularization term, which we'll explore in detail.

## 2. Binary Classification Loss Functions

### 2.1 The Challenge with Zero-One Loss

For binary classification problems with labels $\{+1, -1\}$, the ideal loss function would be the **zero-one loss**:

$$\ell_{0-1}(f(x_n), y_n) = \mathbf{1}(f(x_n) \neq y_n)$$

However, this results in a combinatorial optimization problem that is challenging to solve.

### 2.2 The Hinge Loss Solution

The SVM uses the **hinge loss** as a convex upper bound for the zero-one loss:

$$\ell(t) = \max\{0, 1 - t\}$$

where $t = yf(x) = y(\langle w, x \rangle + b)$

## 3. Understanding the Hinge Loss

### 3.1 Geometric Interpretation

The hinge loss can be expressed in piecewise form:

$$\ell(t) = \begin{cases} 
0 & \text{if } t \geq 1 \\
1 - t & \text{if } t < 1
\end{cases} \quad (12.29)$$

### 3.2 Behavior Analysis

- **$t \geq 1$**: Example is on correct side and outside margin  Loss = 0
- **$0 < t < 1$**: Example is correctly classified but within margin  Positive penalty
- **$t < 0$**: Example is misclassified  Large linear penalty

## 4. SVM Optimization Problem

### 4.1 Unconstrained Formulation

Using hinge loss, the soft margin SVM becomes:

$$\min_{w,b} \frac{1}{2}\|w\|^2 + C \sum_{n=1}^N \max\{0, 1 - y_n(\langle w, x_n \rangle + b)\} \quad (12.31)$$

where:
- $\frac{1}{2}\|w\|^2$ is the **regularization term**
- $C \sum_{n=1}^N \max\{0, 1 - y_n(\langle w, x_n \rangle + b)\}$ is the **loss term**

### 4.2 Constrained Formulation with Slack Variables

The hinge loss minimization:
$$\min_t \max\{0, 1 - t\} \quad (12.32)$$

is equivalent to:
$$\min_{\xi,t} \xi \quad (12.33)$$
$$\text{subject to } \xi \geq 0, \xi \geq 1 - t$$

This leads to the standard soft margin SVM formulation with slack variables.

## 5. Hard Margin vs Soft Margin

### 5.1 Hard Margin Loss

For hard margin SVM, the loss function is:

$$\ell(t) = \begin{cases} 
0 & \text{if } t \geq 1 \\
\infty & \text{if } t < 1
\end{cases} \quad (12.30)$$

This never allows any examples inside the margin.

### 5.2 Connection to Regularization

The key insight is that **margin maximization = regularization**:

$$\text{Maximizing margin} \equiv \text{Minimizing } \frac{1}{2}\|w\|^2$$

## 6. Practical Implementation

```python
import numpy as np
import matplotlib.pyplot as plt

def hinge_loss(t):
    """Compute hinge loss for given t values"""
    return np.maximum(0, 1 - t)

def zero_one_loss(t):
    """Compute zero-one loss for given t values"""
    return (t < 0).astype(int)

# Plot comparison
t = np.linspace(-2, 2, 100)
hinge = hinge_loss(t)
zero_one = zero_one_loss(t)

plt.figure(figsize=(10, 6))
plt.plot(t, hinge, 'b-', linewidth=2, label='Hinge Loss')
plt.plot(t, zero_one, 'r--', linewidth=2, label='Zero-One Loss')
plt.xlabel('t = yf(x)')
plt.ylabel('Loss')
plt.title('Comparison of Loss Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 7. Key Takeaways

1. **Hinge loss** provides a convex approximation to the non-convex zero-one loss
2. The **margin** directly corresponds to the **regularization term**
3. SVM balances between **minimizing training error** and **maximizing margin**
4. The parameter $C$ controls the trade-off between margin and training accuracy

## 8. Mathematical Summary

The complete soft margin SVM optimization problem:

$$\min_{w,b,\xi} \frac{1}{2}\|w\|^2 + C\sum_{n=1}^N \xi_n$$

$$\text{subject to } y_n(\langle w, x_n \rangle + b) \geq 1 - \xi_n, \quad \xi_n \geq 0$$

This formulation elegantly combines:
- **Structural Risk Minimization** (via regularization)
- **Empirical Risk Minimization** (via loss function)
- **Margin Maximization** (geometric interpretation)

````python
import math
import random

class SVM:
    """
    Support Vector Machine implementation with hinge loss
    Based on the soft margin SVM formulation:
    min (1/2)||w||^2 + C * sum(max(0, 1 - y_n * (w^T * x_n + b)))
    """
    
    def __init__(self, C=1.0, learning_rate=0.01, max_iterations=1000, tolerance=1e-6):
        """
        Initialize SVM parameters
        
        Args:
            C: Regularization parameter (trade-off between margin and loss)
            learning_rate: Step size for gradient descent
            max_iterations: Maximum number of training iterations
            tolerance: Convergence threshold
        """
        self.C = C
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.tolerance = tolerance
        self.w = None  # Weight vector
        self.b = 0.0   # Bias term
        self.training_history = []
    
    def _hinge_loss(self, y_true, y_pred):
        """
        Compute hinge loss: max(0, 1 - y_true * y_pred)
        
        Args:
            y_true: True label (+1 or -1)
            y_pred: Predicted value (w^T * x + b)
        
        Returns:
            Hinge loss value
        """
        return max(0, 1 - y_true * y_pred)
    
    def _zero_one_loss(self, y_true, y_pred):
        """
        Compute zero-one loss for comparison
        
        Args:
            y_true: True label (+1 or -1)
            y_pred: Predicted value (w^T * x + b)
        
        Returns:
            0 if correct classification, 1 if incorrect
        """
        predicted_label = 1 if y_pred >= 0 else -1
        return 0 if predicted_label == y_true else 1
    
    def _dot_product(self, x1, x2):
        """Compute dot product of two vectors"""
        return sum(a * b for a, b in zip(x1, x2))
    
    def _vector_add(self, v1, v2):
        """Add two vectors"""
        return [a + b for a, b in zip(v1, v2)]
    
    def _vector_subtract(self, v1, v2):
        """Subtract two vectors"""
        return [a - b for a, b in zip(v1, v2)]
    
    def _scalar_multiply(self, scalar, vector):
        """Multiply vector by scalar"""
        return [scalar * v for v in vector]
    
    def _vector_norm_squared(self, vector):
        """Compute squared L2 norm of vector"""
        return sum(v * v for v in vector)
    
    def _compute_objective(self, X, y):
        """
        Compute the SVM objective function:
        (1/2)||w||^2 + C * sum(hinge_losses)
        """
        if self.w is None:
            return float('inf')
        
        # Regularization term: (1/2)||w||^2
        regularization = 0.5 * self._vector_norm_squared(self.w)
        
        # Loss term: C * sum(hinge_losses)
        total_loss = 0
        for i in range(len(X)):
            y_pred = self._dot_product(self.w, X[i]) + self.b
            total_loss += self._hinge_loss(y[i], y_pred)
        
        return regularization + self.C * total_loss
    
    def _compute_gradients(self, X, y):
        """
        Compute gradients of the objective function w.r.t. w and b
        """
        n_samples = len(X)
        n_features = len(X[0])
        
        # Initialize gradients
        grad_w = [0.0] * n_features
        grad_b = 0.0
        
        for i in range(n_samples):
            y_pred = self._dot_product(self.w, X[i]) + self.b
            margin = y[i] * y_pred
            
            # If example is within margin (hinge loss > 0)
            if margin < 1:
                # Gradient contribution from hinge loss
                grad_w = self._vector_add(grad_w, 
                                        self._scalar_multiply(-self.C * y[i], X[i]))
                grad_b -= self.C * y[i]
        
        # Add regularization gradient for w
        grad_w = self._vector_add(grad_w, self.w)
        
        return grad_w, grad_b
    
    def fit(self, X, y):
        """
        Train the SVM using gradient descent
        
        Args:
            X: Training features (list of lists)
            y: Training labels (list of +1/-1)
        """
        n_samples = len(X)
        n_features = len(X[0])
        
        # Initialize weights randomly
        self.w = [random.uniform(-0.1, 0.1) for _ in range(n_features)]
        self.b = 0.0
        
        prev_objective = float('inf')
        
        for iteration in range(self.max_iterations):
            # Compute gradients
            grad_w, grad_b = self._compute_gradients(X, y)
            
            # Update parameters
            self.w = self._vector_subtract(self.w, 
                                         self._scalar_multiply(self.learning_rate, grad_w))
            self.b -= self.learning_rate * grad_b
            
            # Compute objective value
            current_objective = self._compute_objective(X, y)
            self.training_history.append(current_objective)
            
            # Check for convergence
            if abs(prev_objective - current_objective) < self.tolerance:
                print(f"Converged after {iteration + 1} iterations")
                break
            
            prev_objective = current_objective
            
            if iteration % 100 == 0:
                print(f"Iteration {iteration}: Objective = {current_objective:.6f}")
    
    def predict(self, X):
        """
        Make predictions on new data
        
        Args:
            X: Features to predict (list of lists)
        
        Returns:
            List of predicted labels (+1 or -1)
        """
        if self.w is None:
            raise ValueError("Model must be trained before making predictions")
        
        predictions = []
        for x in X:
            score = self._dot_product(self.w, x) + self.b
            prediction = 1 if score >= 0 else -1
            predictions.append(prediction)
        
        return predictions
    
    def decision_function(self, X):
        """
        Compute decision function values (distances from hyperplane)
        
        Args:
            X: Features (list of lists)
        
        Returns:
            List of decision function values
        """
        if self.w is None:
            raise ValueError("Model must be trained before computing decision function")
        
        return [self._dot_product(self.w, x) + self.b for x in X]
    
    def evaluate(self, X, y):
        """
        Evaluate model performance
        
        Args:
            X: Test features
            y: True labels
        
        Returns:
            Dictionary with various metrics
        """
        predictions = self.predict(X)
        decision_values = self.decision_function(X)
        
        # Compute accuracy
        correct = sum(1 for pred, true in zip(predictions, y) if pred == true)
        accuracy = correct / len(y)
        
        # Compute average hinge loss
        total_hinge_loss = sum(self._hinge_loss(true, pred_val) 
                              for true, pred_val in zip(y, decision_values))
        avg_hinge_loss = total_hinge_loss / len(y)
        
        # Compute average zero-one loss
        total_zero_one_loss = sum(self._zero_one_loss(true, pred_val) 
                                 for true, pred_val in zip(y, decision_values))
        avg_zero_one_loss = total_zero_one_loss / len(y)
        
        # Compute margin (1/||w||)
        w_norm = math.sqrt(self._vector_norm_squared(self.w))
        margin = 1.0 / w_norm if w_norm > 0 else float('inf')
        
        return {
            'accuracy': accuracy,
            'avg_hinge_loss': avg_hinge_loss,
            'avg_zero_one_loss': avg_zero_one_loss,
            'margin': margin,
            'w_norm': w_norm,
            'final_objective': self._compute_objective(X, y)
        }
    
    def get_parameters(self):
        """Return current model parameters"""
        return {'w': self.w.copy() if self.w else None, 'b': self.b}


def generate_linearly_separable_data(n_samples=100, n_features=2, random_seed=42):
    """
    Generate linearly separable synthetic data for testing
    """
    random.seed(random_seed)
    
    X = []
    y = []
    
    for _ in range(n_samples):
        # Generate random point
        x = [random.uniform(-5, 5) for _ in range(n_features)]
        
        # Simple linear decision boundary: x[0] + x[1] > 0
        if n_features >= 2:
            label = 1 if x[0] + x[1] > 0 else -1
        else:
            label = 1 if x[0] > 0 else -1
        
        X.append(x)
        y.append(label)
    
    return X, y


def demonstrate_loss_comparison():
    """Demonstrate the difference between hinge loss and zero-one loss"""
    print("=== Loss Function Comparison ===")
    
    svm = SVM()
    
    # Test different margin values
    test_margins = [-2, -1, -0.5, 0, 0.5, 1, 1.5, 2]
    
    print(f"{'Margin (t)':<10} {'Hinge Loss':<12} {'Zero-One Loss':<15}")
    print("-" * 40)
    
    for t in test_margins:
        hinge = svm._hinge_loss(1, t)  # Assuming y=1
        zero_one = svm._zero_one_loss(1, t)
        print(f"{t:<10} {hinge:<12.2f} {zero_one:<15}")


def main():
    """Demonstrate SVM implementation"""
    print("=== SVM Implementation Demo ===\n")
    
    # Generate synthetic data
    print("Generating synthetic linearly separable data...")
    X_train, y_train = generate_linearly_separable_data(n_samples=200, n_features=2)
    X_test, y_test = generate_linearly_separable_data(n_samples=50, n_features=2, random_seed=123)
    
    print(f"Training samples: {len(X_train)}")
    print(f"Test samples: {len(X_test)}")
    
    # Train SVM with different C values
    C_values = [0.1, 1.0, 10.0]
    
    for C in C_values:
        print(f"\n--- Training SVM with C = {C} ---")
        
        svm = SVM(C=C, learning_rate=0.01, max_iterations=1000)
        svm.fit(X_train, y_train)
        
        # Evaluate on training data
        train_metrics = svm.evaluate(X_train, y_train)
        print(f"Training Accuracy: {train_metrics['accuracy']:.3f}")
        print(f"Training Hinge Loss: {train_metrics['avg_hinge_loss']:.3f}")
        print(f"Margin: {train_metrics['margin']:.3f}")
        
        # Evaluate on test data
        test_metrics = svm.evaluate(X_test, y_test)
        print(f"Test Accuracy: {test_metrics['accuracy']:.3f}")
        print(f"Test Hinge Loss: {test_metrics['avg_hinge_loss']:.3f}")
        
        # Show model parameters
        params = svm.get_parameters()
        print(f"Learned weights: {[f'{w:.3f}' for w in params['w']]}")
        print(f"Learned bias: {params['b']:.3f}")
    
    print("\n")
    demonstrate_loss_comparison()
    
    print("\n=== Key Insights ===")
    print("1. Higher C values lead to lower training error but potentially overfitting")
    print("2. Lower C values emphasize margin maximization (regularization)")
    print("3. Hinge loss provides a convex upper bound for zero-one loss")
    print("4. The margin is inversely proportional to ||w||")


if __name__ == "__main__":
    main()
````

# Dual Support Vector Machine

## 1. Introduction to Dual SVM

The **primal SVM** formulation uses variables $w$ and $b$, where the number of parameters grows linearly with the number of features $D$. The **dual SVM** provides an equivalent optimization problem where the number of parameters scales with the number of training examples $N$ instead.

### Key Advantages of Dual Formulation:
- Independent of the number of features
- Useful when features > training examples
- Enables kernel methods
- Based on convex duality theory

## 2. From Primal to Dual via Lagrange Multipliers

### 2.1 Primal Soft Margin SVM Recap

The primal soft margin SVM optimization problem:

$$\min_{w,b,\xi} \frac{1}{2}\|w\|^2 + C\sum_{n=1}^N \xi_n$$

subject to:
- $y_n(\langle w, x_n \rangle + b) \geq 1 - \xi_n$ for all $n$
- $\xi_n \geq 0$ for all $n$

### 2.2 Lagrangian Formulation

We introduce Lagrange multipliers:
- $\alpha_n \geq 0$ for the margin constraints
- $\gamma_n \geq 0$ for the non-negativity constraints

The **Lagrangian** becomes:

$$L(w, b, \xi, \alpha, \gamma) = \frac{1}{2}\|w\|^2 + C\sum_{n=1}^N \xi_n - \sum_{n=1}^N \alpha_n(y_n(\langle w, x_n \rangle + b) - 1 + \xi_n) - \sum_{n=1}^N \gamma_n \xi_n \quad (12.34)$$

## 3. Optimality Conditions

### 3.1 Partial Derivatives

Taking partial derivatives with respect to primal variables:

$$\frac{\partial L}{\partial w} = w^T - \sum_{n=1}^N \alpha_n y_n x_n^T \quad (12.35)$$

$$\frac{\partial L}{\partial b} = -\sum_{n=1}^N \alpha_n y_n \quad (12.36)$$

$$\frac{\partial L}{\partial \xi_n} = C - \alpha_n - \gamma_n \quad (12.37)$$

### 3.2 Setting Derivatives to Zero

From equation (12.35) = 0:
$$w = \sum_{n=1}^N \alpha_n y_n x_n \quad (12.38)$$

From equation (12.36) = 0:
$$\sum_{n=1}^N \alpha_n y_n = 0$$

From equation (12.37) = 0:
$$C = \alpha_n + \gamma_n$$

## 4. The Representer Theorem

### 4.1 Key Insight

Equation (12.38) is a specific instance of the **representer theorem**:

$$w = \sum_{n=1}^N \alpha_n y_n x_n$$

**Interpretation**: The optimal weight vector is a linear combination of the training examples.

### 4.2 Support Vectors

- **Support vectors**: Examples $x_n$ where $\alpha_n > 0$
- **Non-support vectors**: Examples where $\alpha_n = 0$ (don't contribute to $w$)
- Only support vectors "support" the decision hyperplane

## 5. Deriving the Dual Problem

### 5.1 Substituting Back into Lagrangian

Substituting $w = \sum_{n=1}^N \alpha_n y_n x_n$ into (12.34):

$$D(\xi, \alpha, \gamma) = \frac{1}{2}\sum_{i=1}^N \sum_{j=1}^N y_i y_j \alpha_i \alpha_j \langle x_i, x_j \rangle - \sum_{i=1}^N y_i \alpha_i \sum_{j=1}^N y_j \alpha_j \langle x_j, x_i \rangle + \ldots \quad (12.39)$$

### 5.2 Simplification Steps

1. The term involving $b$ vanishes due to $\sum_{n=1}^N \alpha_n y_n = 0$
2. Inner products are symmetric: $\langle x_i, x_j \rangle = \langle x_j, x_i \rangle$
3. Many terms cancel out

### 5.3 Final Dual Formulation

After algebraic manipulation, the dual problem becomes:

$$\max_{\alpha} \sum_{n=1}^N \alpha_n - \frac{1}{2}\sum_{i=1}^N \sum_{j=1}^N \alpha_i \alpha_j y_i y_j \langle x_i, x_j \rangle$$

subject to:
- $\sum_{n=1}^N \alpha_n y_n = 0$
- $0 \leq \alpha_n \leq C$ for all $n$

## 6. Complementary Slackness Conditions

### 6.1 KKT Conditions

For the optimal solution, the following must hold:

1. **Primal feasibility**: 
   - $y_n(\langle w, x_n \rangle + b) \geq 1 - \xi_n$
   - $\xi_n \geq 0$

2. **Dual feasibility**: 
   - $\alpha_n \geq 0$
   - $\gamma_n \geq 0$

3. **Complementary slackness**:
   - $\alpha_n(y_n(\langle w, x_n \rangle + b) - 1 + \xi_n) = 0$
   - $\gamma_n \xi_n = 0$

### 6.2 Classification of Training Points

Based on the KKT conditions:

1. **$\alpha_n = 0$**: Non-support vectors (correctly classified, outside margin)
2. **$0 < \alpha_n < C$**: Support vectors on the margin ($\xi_n = 0$)
3. **$\alpha_n = C$**: Support vectors inside margin or misclassified ($\xi_n > 0$)

## 7. Advantages of Dual Formulation

### 7.1 Computational Benefits

```python
# Primal problem complexity: O(D) parameters
# where D = number of features

# Dual problem complexity: O(N) parameters  
# where N = number of training examples

# Beneficial when D >> N (high-dimensional data, few samples)
```

### 7.2 Kernel Methods

The dual formulation only requires inner products $\langle x_i, x_j \rangle$, enabling:

$$\langle x_i, x_j \rangle \rightarrow k(x_i, x_j)$$

where $k(\cdot, \cdot)$ is a kernel function.

## 8. Implementation Considerations

### 8.1 Quadratic Programming

The dual SVM is a **quadratic programming** problem:
- Quadratic objective function
- Linear equality constraint
- Box constraints on variables

### 8.2 Solving Methods

1. **Sequential Minimal Optimization (SMO)**
2. **Interior point methods**
3. **Active set methods**

## 9. Prediction in Dual Form

### 9.1 Decision Function

For a new point $x$:

$$f(x) = \sum_{n=1}^N \alpha_n y_n \langle x_n, x \rangle + b$$

Only support vectors ($\alpha_n > 0$) contribute to the sum.

### 9.2 Bias Term Recovery

The bias $b$ can be recovered from any support vector on the margin:

$$b = y_n - \sum_{m=1}^N \alpha_m y_m \langle x_m, x_n \rangle$$

where $0 < \alpha_n < C$ (support vector on margin).

## 10. Summary

### Key Takeaways

1. **Dual formulation** scales with number of examples, not features
2. **Representer theorem** shows optimal $w$ lies in span of training data
3. **Support vectors** are the critical examples that define the hyperplane
4. **KKT conditions** characterize the optimal solution
5. **Kernel methods** become natural in dual formulation

### Mathematical Elegance

The transformation from primal to dual reveals the geometric structure of SVM:
- Decision boundary determined by support vectors only
- Margin maximization becomes quadratic optimization
- Sparsity naturally emerges through complementary slackness

```python
# Dual SVM prediction depends only on support vectors
def predict_dual(x_new, support_vectors, alphas, labels, b):
    decision_value = b
    for i in range(len(support_vectors)):
        decision_value += alphas[i] * labels[i] * dot_product(support_vectors[i], x_new)
    return sign(decision_value)
```

````python
import math
import random

class DualSVM:
    """
    Dual Support Vector Machine implementation
    
    Solves the dual optimization problem:
    max (_i) - (1/2)_i _j _i _j y_i y_j <x_i, x_j>
    subject to: (_i y_i) = 0, 0  _i  C
    
    Uses Sequential Minimal Optimization (SMO) algorithm
    """
    
    def __init__(self, C=1.0, tolerance=1e-3, max_passes=5, kernel='linear'):
        """
        Initialize Dual SVM
        
        Args:
            C: Regularization parameter
            tolerance: Numerical tolerance for convergence
            max_passes: Maximum passes through dataset
            kernel: Kernel type ('linear', 'polynomial', 'rbf')
        """
        self.C = C
        self.tolerance = tolerance
        self.max_passes = max_passes
        self.kernel_type = kernel
        
        # Model parameters
        self.alphas = None      # Lagrange multipliers
        self.b = 0.0           # Bias term
        self.X = None          # Training data
        self.y = None          # Training labels
        self.support_vectors = []
        self.support_labels = []
        self.support_alphas = []
        
        # Kernel parameters
        self.gamma = 1.0       # For RBF kernel
        self.degree = 3        # For polynomial kernel
        self.coef0 = 0.0       # For polynomial kernel
        
        # Training statistics
        self.training_history = []
    
    def kernel(self, x1, x2):
        """
        Compute kernel function k(x1, x2)
        """
        if self.kernel_type == 'linear':
            return self._dot_product(x1, x2)
        elif self.kernel_type == 'polynomial':
            return (self.gamma * self._dot_product(x1, x2) + self.coef0) ** self.degree
        elif self.kernel_type == 'rbf':
            # RBF (Gaussian) kernel: exp(-gamma * ||x1 - x2||^2)
            diff = self._vector_subtract(x1, x2)
            norm_sq = self._vector_norm_squared(diff)
            return math.exp(-self.gamma * norm_sq)
        else:
            raise ValueError(f"Unknown kernel type: {self.kernel_type}")
    
    def _dot_product(self, x1, x2):
        """Compute dot product of two vectors"""
        return sum(a * b for a, b in zip(x1, x2))
    
    def _vector_subtract(self, v1, v2):
        """Subtract two vectors"""
        return [a - b for a, b in zip(v1, v2)]
    
    def _vector_norm_squared(self, vector):
        """Compute squared L2 norm of vector"""
        return sum(v * v for v in vector)
    
    def _compute_dual_objective(self):
        """
        Compute dual objective function value:
        (_i) - (1/2)_i _j _i _j y_i y_j k(x_i, x_j)
        """
        n = len(self.alphas)
        
        # First term: (_i)
        first_term = sum(self.alphas)
        
        # Second term: (1/2)_i _j _i _j y_i y_j k(x_i, x_j)
        second_term = 0.0
        for i in range(n):
            for j in range(n):
                if self.alphas[i] > 0 and self.alphas[j] > 0:
                    kernel_val = self.kernel(self.X[i], self.X[j])
                    second_term += self.alphas[i] * self.alphas[j] * self.y[i] * self.y[j] * kernel_val
        
        return first_term - 0.5 * second_term
    
    def _decision_function_single(self, x):
        """
        Compute decision function for single point:
        f(x) = (_i y_i k(x_i, x)) + b
        """
        result = 0.0
        for i in range(len(self.X)):
            if self.alphas[i] > 0:  # Only support vectors contribute
                result += self.alphas[i] * self.y[i] * self.kernel(self.X[i], x)
        return result + self.b
    
    def _compute_error(self, i):
        """Compute prediction error for example i"""
        prediction = self._decision_function_single(self.X[i])
        return prediction - self.y[i]
    
    def _select_second_alpha(self, i1, E1):
        """
        Select second alpha using heuristic:
        Choose alpha that maximizes |E1 - E2|
        """
        i2 = -1
        max_delta_E = 0
        E2 = 0
        
        # Look for alpha with maximum |E1 - E2|
        for i in range(len(self.alphas)):
            if 0 < self.alphas[i] < self.C:
                E_temp = self._compute_error(i)
                delta_E = abs(E1 - E_temp)
                if delta_E > max_delta_E:
                    max_delta_E = delta_E
                    i2 = i
                    E2 = E_temp
        
        # If no suitable alpha found, select randomly
        if i2 == -1:
            i2 = i1
            while i2 == i1:
                i2 = random.randint(0, len(self.alphas) - 1)
            E2 = self._compute_error(i2)
        
        return i2, E2
    
    def _clip_alpha(self, alpha, L, H):
        """Clip alpha to be within bounds [L, H]"""
        if alpha > H:
            return H
        elif alpha < L:
            return L
        else:
            return alpha
    
    def _update_alphas_smo(self, i1, i2):
        """
        Update alphas using SMO algorithm
        Returns True if alphas were updated
        """
        if i1 == i2:
            return False
        
        alpha1_old = self.alphas[i1]
        alpha2_old = self.alphas[i2]
        y1, y2 = self.y[i1], self.y[i2]
        
        # Compute bounds L and H
        if y1 != y2:
            L = max(0, alpha2_old - alpha1_old)
            H = min(self.C, self.C + alpha2_old - alpha1_old)
        else:
            L = max(0, alpha1_old + alpha2_old - self.C)
            H = min(self.C, alpha1_old + alpha2_old)
        
        if L == H:
            return False
        
        # Compute kernel values
        k11 = self.kernel(self.X[i1], self.X[i1])
        k12 = self.kernel(self.X[i1], self.X[i2])
        k22 = self.kernel(self.X[i2], self.X[i2])
        
        # Compute eta (second derivative of objective function)
        eta = k11 + k22 - 2 * k12
        
        if eta <= 0:
            return False
        
        # Compute errors
        E1 = self._compute_error(i1)
        E2 = self._compute_error(i2)
        
        # Compute new alpha2
        alpha2_new = alpha2_old + y2 * (E1 - E2) / eta
        alpha2_new = self._clip_alpha(alpha2_new, L, H)
        
        # Check if change is significant
        if abs(alpha2_new - alpha2_old) < 1e-5:
            return False
        
        # Compute new alpha1
        alpha1_new = alpha1_old + y1 * y2 * (alpha2_old - alpha2_new)
        
        # Update alphas
        self.alphas[i1] = alpha1_new
        self.alphas[i2] = alpha2_new
        
        # Update bias term
        b1 = E1 + y1 * (alpha1_new - alpha1_old) * k11 + y2 * (alpha2_new - alpha2_old) * k12 + self.b
        b2 = E2 + y1 * (alpha1_new - alpha1_old) * k12 + y2 * (alpha2_new - alpha2_old) * k22 + self.b
        
        if 0 < alpha1_new < self.C:
            self.b = b1
        elif 0 < alpha2_new < self.C:
            self.b = b2
        else:
            self.b = (b1 + b2) / 2.0
        
        return True
    
    def _examine_example(self, i2):
        """
        Examine example i2 and try to find a pair to optimize
        """
        y2 = self.y[i2]
        alpha2 = self.alphas[i2]
        E2 = self._compute_error(i2)
        r2 = E2 * y2
        
        # Check KKT conditions
        if ((r2 < -self.tolerance and alpha2 < self.C) or 
            (r2 > self.tolerance and alpha2 > 0)):
            
            # Count non-bound alphas
            non_bound_count = sum(1 for alpha in self.alphas if 0 < alpha < self.C)
            
            if non_bound_count > 1:
                # Select second alpha heuristically
                i1, E1 = self._select_second_alpha(i2, E2)
                if self._update_alphas_smo(i1, i2):
                    return True
            
            # Try all non-bound alphas as second choice
            start_idx = random.randint(0, len(self.alphas) - 1)
            for i in range(len(self.alphas)):
                i1 = (start_idx + i) % len(self.alphas)
                if 0 < self.alphas[i1] < self.C:
                    if self._update_alphas_smo(i1, i2):
                        return True
            
            # Try all alphas as second choice
            start_idx = random.randint(0, len(self.alphas) - 1)
            for i in range(len(self.alphas)):
                i1 = (start_idx + i) % len(self.alphas)
                if self._update_alphas_smo(i1, i2):
                    return True
        
        return False
    
    def fit(self, X, y):
        """
        Train the dual SVM using SMO algorithm
        
        Args:
            X: Training features (list of lists)
            y: Training labels (list of +1/-1)
        """
        self.X = [x[:] for x in X]  # Deep copy
        self.y = y[:]               # Copy
        n_samples = len(X)
        
        # Initialize alphas
        self.alphas = [0.0] * n_samples
        self.b = 0.0
        
        passes = 0
        num_changed = 0
        examine_all = True
        
        print(f"Starting SMO algorithm with {n_samples} samples...")
        
        while (num_changed > 0 or examine_all) and passes < self.max_passes:
            num_changed = 0
            
            if examine_all:
                # Examine all examples
                for i in range(n_samples):
                    if self._examine_example(i):
                        num_changed += 1
            else:
                # Examine non-bound examples
                for i in range(n_samples):
                    if 0 < self.alphas[i] < self.C:
                        if self._examine_example(i):
                            num_changed += 1
            
            if examine_all:
                examine_all = False
            elif num_changed == 0:
                examine_all = True
            
            passes += 1
            
            # Record training progress
            dual_obj = self._compute_dual_objective()
            self.training_history.append(dual_obj)
            
            if passes % max(1, self.max_passes // 10) == 0:
                support_count = sum(1 for alpha in self.alphas if alpha > 1e-6)
                print(f"Pass {passes}: {num_changed} changed, {support_count} support vectors, "
                      f"dual objective = {dual_obj:.6f}")
        
        print(f"SMO converged after {passes} passes")
        
        # Extract support vectors
        self._extract_support_vectors()
        
        # Compute primal weights if using linear kernel
        if self.kernel_type == 'linear':
            self._compute_primal_weights()
    
    def _extract_support_vectors(self):
        """Extract support vectors (examples with  > 0)"""
        self.support_vectors = []
        self.support_labels = []
        self.support_alphas = []
        
        for i in range(len(self.alphas)):
            if self.alphas[i] > 1e-6:  # Consider as support vector
                self.support_vectors.append(self.X[i])
                self.support_labels.append(self.y[i])
                self.support_alphas.append(self.alphas[i])
    
    def _compute_primal_weights(self):
        """
        Compute primal weights w = (_i y_i x_i) for linear kernel
        This demonstrates the representer theorem
        """
        if not self.X or self.kernel_type != 'linear':
            return None
        
        n_features = len(self.X[0])
        self.w = [0.0] * n_features
        
        for i in range(len(self.alphas)):
            if self.alphas[i] > 0:
                for j in range(n_features):
                    self.w[j] += self.alphas[i] * self.y[i] * self.X[i][j]
        
        return self.w
    
    def predict(self, X):
        """
        Make predictions on new data
        
        Args:
            X: Features to predict (list of lists)
        
        Returns:
            List of predicted labels (+1 or -1)
        """
        predictions = []
        for x in X:
            decision_value = self._decision_function_single(x)
            prediction = 1 if decision_value >= 0 else -1
            predictions.append(prediction)
        return predictions
    
    def decision_function(self, X):
        """
        Compute decision function values for input data
        
        Args:
            X: Features (list of lists)
        
        Returns:
            List of decision function values
        """
        return [self._decision_function_single(x) for x in X]
    
    def get_support_vectors(self):
        """Return support vectors and related information"""
        return {
            'support_vectors': self.support_vectors,
            'support_labels': self.support_labels,
            'support_alphas': self.support_alphas,
            'n_support': len(self.support_vectors)
        }
    
    def get_dual_coefficients(self):
        """Return dual coefficients (_i y_i for support vectors)"""
        return [alpha * label for alpha, label in zip(self.support_alphas, self.support_labels)]
    
    def evaluate(self, X, y):
        """Evaluate model performance"""
        predictions = self.predict(X)
        decision_values = self.decision_function(X)
        
        # Compute accuracy
        correct = sum(1 for pred, true in zip(predictions, y) if pred == true)
        accuracy = correct / len(y)
        
        # Compute average hinge loss
        total_hinge_loss = 0
        for i, (true_label, decision_val) in enumerate(zip(y, decision_values)):
            hinge_loss = max(0, 1 - true_label * decision_val)
            total_hinge_loss += hinge_loss
        avg_hinge_loss = total_hinge_loss / len(y)
        
        return {
            'accuracy': accuracy,
            'avg_hinge_loss': avg_hinge_loss,
            'n_support_vectors': len(self.support_vectors),
            'dual_objective': self._compute_dual_objective()
        }


def compare_primal_dual_svm():
    """Compare primal and dual SVM implementations"""
    from math import sqrt
    
    print("=== Comparing Primal and Dual SVM ===\n")
    
    # Generate simple 2D dataset
    random.seed(42)
    X = []
    y = []
    
    # Generate linearly separable data
    for _ in range(100):
        x1 = random.uniform(-3, 3)
        x2 = random.uniform(-3, 3)
        X.append([x1, x2])
        # Decision boundary: x1 + x2 > 0
        y.append(1 if x1 + x2 > 0.5 else -1)
    
    print(f"Dataset: {len(X)} samples, {len(X[0])} features")
    
    # Train dual SVM
    dual_svm = DualSVM(C=1.0, kernel='linear', max_passes=10)
    dual_svm.fit(X, y)
    
    # Get results
    support_info = dual_svm.get_support_vectors()
    dual_coeffs = dual_svm.get_dual_coefficients()
    
    print(f"\nDual SVM Results:")
    print(f"Number of support vectors: {support_info['n_support']}")
    print(f"Support vector indices: {[i for i, alpha in enumerate(dual_svm.alphas) if alpha > 1e-6]}")
    print(f"Bias term (b): {dual_svm.b:.4f}")
    
    if hasattr(dual_svm, 'w') and dual_svm.w:
        print(f"Primal weights (w): [{dual_svm.w[0]:.4f}, {dual_svm.w[1]:.4f}]")
        w_norm = sqrt(sum(w*w for w in dual_svm.w))
        print(f"Margin (1/||w||): {1/w_norm:.4f}")
    
    # Show some support vectors
    print(f"\nSupport Vectors (first 5):")
    for i in range(min(5, len(support_info['support_vectors']))):
        sv = support_info['support_vectors'][i]
        alpha = support_info['support_alphas'][i]
        label = support_info['support_labels'][i]
        print(f"  SV {i+1}: x=[{sv[0]:.2f}, {sv[1]:.2f}], y={label:+d}, ={alpha:.4f}")
    
    # Evaluate performance
    metrics = dual_svm.evaluate(X, y)
    print(f"\nPerformance:")
    print(f"Training accuracy: {metrics['accuracy']:.3f}")
    print(f"Average hinge loss: {metrics['avg_hinge_loss']:.3f}")
    print(f"Dual objective: {metrics['dual_objective']:.3f}")
    
    # Test predictions
    test_points = [[1, 1], [-1, -1], [0, 2], [2, -1]]
    predictions = dual_svm.predict(test_points)
    decision_vals = dual_svm.decision_function(test_points)
    
    print(f"\nTest Predictions:")
    for i, (point, pred, dec_val) in enumerate(zip(test_points, predictions, decision_vals)):
        print(f"  Point {point}: prediction={pred:+d}, decision_value={dec_val:.3f}")


def demonstrate_kernel_svm():
    """Demonstrate SVM with different kernels"""
    print("\n=== Kernel SVM Demonstration ===\n")
    
    # Generate non-linearly separable data (XOR-like pattern)
    random.seed(123)
    X = []
    y = []
    
    # Generate XOR pattern
    for _ in range(200):
        x1 = random.uniform(-2, 2)
        x2 = random.uniform(-2, 2)
        X.append([x1, x2])
        # XOR pattern: positive if signs are different
        y.append(1 if x1 * x2 < 0 else -1)
    
    print(f"Non-linear dataset: {len(X)} samples")
    
    # Test different kernels
    kernels = ['linear', 'polynomial', 'rbf']
    
    for kernel_type in kernels:
        print(f"\n--- {kernel_type.upper()} Kernel ---")
        
        svm = DualSVM(C=1.0, kernel=kernel_type, max_passes=5)
        if kernel_type == 'rbf':
            svm.gamma = 0.5
        elif kernel_type == 'polynomial':
            svm.degree = 2
        
        svm.fit(X, y)
        
        metrics = svm.evaluate(X, y)
        support_info = svm.get_support_vectors()
        
        print(f"Accuracy: {metrics['accuracy']:.3f}")
        print(f"Support vectors: {support_info['n_support']}")
        print(f"Dual objective: {metrics['dual_objective']:.3f}")


def main():
    """Main demonstration function"""
    print("=== Dual SVM Implementation ===\n")
    
    # Compare primal and dual approaches
    compare_primal_dual_svm()
    
    # Demonstrate kernel methods
    demonstrate_kernel_svm()
    
    print("\n=== Key Insights ===")
    print("1. Dual formulation scales with number of examples, not features")
    print("2. Only support vectors ( > 0) contribute to the decision function")
    print("3. Representer theorem: w = (_i y_i x_i)")
    print("4. Kernel methods work naturally in dual formulation")
    print("5. SMO algorithm efficiently solves the dual quadratic programming problem")


if __name__ == "__main__":
    main()
````

## 12.3.2 Dual SVM: Convex Hull View

An alternative geometric approach to deriving the dual Support Vector Machine (SVM) involves the concept of **convex hulls**. Consider the set of training examples $ \{ (x_1, y_1), \ldots, (x_N, y_N) \} $, where $ x_n $ are the feature vectors and $ y_n \in \{+1, -1\} $ are the labels. We aim to build the smallest convex set that contains all examples of the same class, known as the **convex hull**, as illustrated in Figure 12.9.

### Understanding Convex Hulls

#### Convex Combination of Points

To build intuition, consider a convex combination of points. For two points $ x_1 $ and $ x_2 $ with non-negative weights $ \alpha_1, \alpha_2 \geq 0 $ such that $ \alpha_1 + \alpha_2 = 1 $, the expression $ \alpha_1 x_1 + \alpha_2 x_2 $ describes all points on the line segment between $ x_1 $ and $ x_2 $.

Adding a third point $ x_3 $ with weight $ \alpha_3 \geq 0 $ such that $ \sum_{n=1}^3 \alpha_n = 1 $, the convex combination $ \alpha_1 x_1 + \alpha_2 x_2 + \alpha_3 x_3 $ spans a two-dimensional areaa triangle with vertices $ x_1 $, $ x_2 $, and $ x_3 $. As more points are added, and the number of points exceeds the dimensionality of the space, some points lie inside the convex hull, as shown in Figure 12.9(a).

#### Definition of the Convex Hull

In general, the convex hull of a set of points $ X = \{ x_1, \ldots, x_N \} $ is defined as the set of all convex combinations of those points. For non-negative weights $ \alpha_n \geq 0 $ that sum to 1, the convex hull is:

$$
\text{conv}(X) = \left\{ \sum_{n=1}^N \alpha_n x_n \ \Big| \ \sum_{n=1}^N \alpha_n = 1, \ \alpha_n \geq 0 \ \text{for all} \ n = 1, \ldots, N \right\}. \tag{12.43}
$$

### Convex Hulls for Positive and Negative Classes

If the positive ($ y_n = +1 $) and negative ($ y_n = -1 $) classes are linearly separable, their convex hulls do not overlap. We form two convex hulls:
- One for the positive examples ($ y_n = +1 $).
- One for the negative examples ($ y_n = -1 $).

We select:
- A point $ c $ in the convex hull of the positive examples, closest to the negative class.
- A point $ d $ in the convex hull of the negative examples, closest to the positive class.

This is illustrated in Figure 12.9(b), where the positive examples are in blue and the negative examples are in orange. The distance between the two convex hulls is the length of the difference vector $ c - d $.

Define the difference vector:

$$
w := c - d. \tag{12.44}
$$

Choosing $ c $ and $ d $ to be the closest points between the two convex hulls is equivalent to minimizing the norm of $ w $. Thus, the optimization problem becomes:

$$
\arg \min_w \|w\| = \arg \min_w \frac{1}{2} \|w\|^2. \tag{12.45}
$$

### Expressing Points in the Convex Hulls

Since $ c $ lies in the convex hull of the positive examples, it can be written as a convex combination of those examples:

$$
c = \sum_{n: y_n = +1} \alpha_n^+ x_n, \tag{12.46}
$$

where $ \alpha_n^+ \geq 0 $ and $ \sum_{n: y_n = +1} \alpha_n^+ = 1 $. The notation $ n: y_n = +1 $ indicates the set of indices where $ y_n = +1 $.

Similarly, for the negative examples:

$$
d = \sum_{n: y_n = -1} \alpha_n^- x_n, \tag{12.47}
$$

where $ \alpha_n^- \geq 0 $ and $ \sum_{n: y_n = -1} \alpha_n^- = 1 $.

### Formulating the Optimization Problem

Substitute Equations (12.46) and (12.47) into Equation (12.44):

$$
w = c - d = \sum_{n: y_n = +1} \alpha_n^+ x_n - \sum_{n: y_n = -1} \alpha_n^- x_n.
$$

The objective function (Equation 12.45) becomes:

$$
\min_{\alpha} \frac{1}{2} \left\| \sum_{n: y_n = +1} \alpha_n^+ x_n - \sum_{n: y_n = -1} \alpha_n^- x_n \right\|^2. \tag{12.48}
$$

Let $ \alpha $ be the vector of all coefficients, concatenating $ \alpha^+ $ (for positive examples) and $ \alpha^- $ (for negative examples). The convex hull constraints require that the coefficients for each class sum to 1:

$$
\sum_{n: y_n = +1} \alpha_n^+ = 1 \quad \text{and} \quad \sum_{n: y_n = -1} \alpha_n^- = 1. \tag{12.49}
$$

This implies a combined constraint:

$$
\sum_{n=1}^N y_n \alpha_n = 0. \tag{12.50}
$$

To derive this, consider:

$$
\sum_{n=1}^N y_n \alpha_n = \sum_{n: y_n = +1} (+1) \alpha_n^+ + \sum_{n: y_n = -1} (-1) \alpha_n^-, \tag{12.51a}
$$

$$
= \sum_{n: y_n = +1} \alpha_n^+ - \sum_{n: y_n = -1} \alpha_n^- = 1 - 1 = 0. \tag{12.51b}
$$

### Connection to the Dual Hard Margin SVM

The optimization problem defined by the objective function (Equation 12.48), the constraint (Equation 12.50), and the condition $ \alpha_n \geq 0 $, forms a constrained convex optimization problem. This problem is equivalent to the dual form of the hard margin SVM (Bennett and Bredensteiner, 2000a). The vector $ w $ represents the weight vector of the separating hyperplane, and minimizing its norm maximizes the margin between the two convex hulls.

### Remark: Soft Margin via Reduced Hull

To extend this to the soft margin SVM, we consider the **reduced hull**. The reduced hull is similar to the convex hull but imposes an upper bound on the coefficients $ \alpha_n $. This upper bound limits the influence of any single example, effectively shrinking the convex hull to a smaller volume (Bennett and Bredensteiner, 2000b). This modification allows the SVM to handle non-separable data by introducing slack variables, leading to the soft margin dual formulation. $ \diamond $

````python
import math
import matplotlib.pyplot as plt
from itertools import combinations

class HardMarginSVM:
    """
    Hard Margin SVM implementation using the convex hull perspective in 2D.
    """
    
    def __init__(self):
        self.w = None  # Weight vector
        self.b = None  # Bias term
        self.support_vectors = None
    
    def dot_product(self, x1, x2):
        """Compute dot product of two 2D vectors."""
        return x1[0] * x2[0] + x1[1] * x2[1]
    
    def norm(self, x):
        """Compute the Euclidean norm of a 2D vector."""
        return math.sqrt(self.dot_product(x, x))
    
    def subtract(self, x1, x2):
        """Subtract two 2D vectors."""
        return (x1[0] - x2[0], x1[1] - x2[1])
    
    def add(self, x1, x2):
        """Add two 2D vectors."""
        return (x1[0] + x2[0], x1[1] + x2[1])
    
    def scale(self, x, scalar):
        """Scale a 2D vector by a scalar."""
        return (x[0] * scalar, x[1] * scalar)
    
    def distance_point_to_point(self, p1, p2):
        """Compute the Euclidean distance between two 2D points."""
        return self.norm(self.subtract(p1, p2))
    
    def convex_hull_2d(self, points):
        """
        Compute the vertices of the convex hull in 2D using a simple algorithm.
        Returns the ordered list of points forming the hull.
        """
        if len(points) < 3:
            return points
        
        # Find the leftmost point
        points = sorted(points, key=lambda p: (p[0], p[1]))
        hull = []
        
        # Upper hull
        for p in points:
            while len(hull) >= 2 and self.cross_product(hull[-2], hull[-1], p) <= 0:
                hull.pop()
            hull.append(p)
        
        # Lower hull
        t = len(hull)
        for p in reversed(points[:-1]):
            while len(hull) > t and self.cross_product(hull[-2], hull[-1], p) <= 0:
                hull.pop()
            hull.append(p)
        
        return hull[:-1]  # Remove the last point (duplicate of the first)
    
    def cross_product(self, p1, p2, p3):
        """
        Compute the cross product (p2 - p1) x (p3 - p1) to determine orientation.
        Returns > 0 if counterclockwise, < 0 if clockwise, 0 if collinear.
        """
        return (p2[0] - p1[0]) * (p3[1] - p1[1]) - (p2[1] - p1[1]) * (p3[0] - p1[0])
    
    def closest_points_between_hulls(self, hull_pos, hull_neg):
        """
        Find the closest pair of points between two convex hulls.
        Returns the points (c, d) and their indices in the original hulls.
        """
        min_dist = float('inf')
        c, d = None, None
        c_idx, d_idx = -1, -1
        
        for i, p in enumerate(hull_pos):
            for j, q in enumerate(hull_neg):
                dist = self.distance_point_to_point(p, q)
                if dist < min_dist:
                    min_dist = dist
                    c, d = p, q
                    c_idx, d_idx = i, j
        
        return c, d, c_idx, d_idx, min_dist
    
    def fit(self, X, y):
        """
        Fit a hard margin SVM using the convex hull approach.
        
        Parameters:
        -----------
        X : list of tuples
            2D data points [(x1, x2), ...]
        y : list of int
            Labels (+1 or -1)
        """
        # Separate positive and negative examples
        pos_points = [X[i] for i in range(len(X)) if y[i] == 1]
        neg_points = [X[i] for i in range(len(X)) if y[i] == -1]
        
        # Compute convex hulls
        hull_pos = self.convex_hull_2d(pos_points)
        hull_neg = self.convex_hull_2d(neg_points)
        
        # Find closest points c and d between the hulls
        c, d, c_idx, d_idx, min_dist = self.closest_points_between_hulls(hull_pos, hull_neg)
        
        # In this simplified implementation, c and d are the support vectors
        # Map back to original indices in X
        pos_indices = [i for i in range(len(X)) if y[i] == 1]
        neg_indices = [i for i in range(len(X)) if y[i] == -1]
        c_original_idx = pos_indices[pos_points.index(hull_pos[c_idx])]
        d_original_idx = neg_indices[neg_points.index(hull_neg[d_idx])]
        self.support_vectors = [(X[c_original_idx], 1), (X[d_original_idx], -1)]
        
        # Compute w = c - d
        self.w = self.subtract(c, d)
        
        # Compute bias b using the fact that the hyperplane bisects c and d
        # For a point on the hyperplane: w^T x + b = 0
        # Midpoint of c and d should lie on the hyperplane
        midpoint = self.scale(self.add(c, d), 0.5)
        self.b = -self.dot_product(self.w, midpoint)
    
    def predict(self, x):
        """
        Predict the class of a 2D point.
        
        Parameters:
        -----------
        x : tuple
            2D point (x1, x2)
            
        Returns:
        --------
        int : Predicted class (+1 or -1)
        """
        return 1 if self.dot_product(self.w, x) + self.b > 0 else -1

def plot_svm(X, y, svm):
    """Plot the data points, convex hulls, separating hyperplane, and margins."""
    # Separate positive and negative points
    pos_points = [X[i] for i in range(len(X)) if y[i] == 1]
    neg_points = [X[i] for i in range(len(X)) if y[i] == -1]
    
    # Compute convex hulls
    hull_pos = svm.convex_hull_2d(pos_points)
    hull_neg = svm.convex_hull_2d(neg_points)
    
    # Find closest points c and d
    c, d, _, _, _ = svm.closest_points_between_hulls(hull_pos, hull_neg)
    
    # Plot
    plt.figure(figsize=(8, 6))
    
    # Plot points
    plt.scatter([p[0] for p in pos_points], [p[1] for p in pos_points], color='blue', label='Positive (+1)', s=100)
    plt.scatter([p[0] for p in neg_points], [p[1] for p in neg_points], color='orange', label='Negative (-1)', s=100)
    
    # Plot convex hulls
    hull_pos.append(hull_pos[0])  # Close the loop
    hull_neg.append(hull_neg[0])
    plt.plot([p[0] for p in hull_pos], [p[1] for p in hull_pos], 'b--', label='Positive Convex Hull')
    plt.plot([p[0] for p in hull_neg], [p[1] for p in hull_neg], 'orange', linestyle='--', label='Negative Convex Hull')
    
    # Plot points c and d
    plt.scatter([c[0]], [c[1]], color='blue', marker='*', s=200, label='Point c')
    plt.scatter([d[0]], [d[1]], color='orange', marker='*', s=200, label='Point d')
    
    # Plot the separating hyperplane w^T x + b = 0
    # In 2D: w1 x1 + w2 x2 + b = 0 => x2 = (-w1 x1 - b) / w2
    x_range = [min([p[0] for p in X]) - 1, max([p[0] for p in X]) + 1]
    x1 = x_range
    w1, w2 = svm.w
    b = svm.b
    x2 = [(-w1 * x - b) / w2 for x in x1]
    plt.plot(x1, x2, 'k-', label='Separating Hyperplane')
    
    # Plot margins (w^T x + b = 1)
    x2_pos = [(-w1 * x - b + 1) / w2 for x in x1]
    x2_neg = [(-w1 * x - b - 1) / w2 for x in x1]
    plt.plot(x1, x2_pos, 'k--', label='Margins')
    plt.plot(x1, x2_neg, 'k--')
    
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Hard Margin SVM: Convex Hull Perspective')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

if __name__ == "__main__":
    print("Hard Margin SVM Implementation in Core Python (Convex Hull Perspective)")
    print("=" * 60)
    
    # Simple 2D dataset (linearly separable)
    X = [(1, 1), (2, 2), (3, 1), (4, 2),  # Positive class
         (6, 6), (7, 5), (8, 6), (9, 5)]   # Negative class
    y = [1, 1, 1, 1, -1, -1, -1, -1]
    
    # Initialize and train the SVM
    svm = HardMarginSVM()
    svm.fit(X, y)
    
    # Print results
    print(f"Weight vector w: {svm.w}")
    print(f"Bias b: {round(svm.b, 2)}")
    print("Support Vectors:")
    for sv, label in svm.support_vectors:
        print(f"Point: {sv}, Label: {label}")
    
    # Plot the results
    plot_svm(X, y, svm)
    
    print("\n=== Demonstration completed ===")
````

# 12.4 Kernels

Consider the dual SVM formulation (Equation 12.41). In this formulation, the inner product appears only between pairs of examples $ x_i $ and $ x_j $, and not between the examples and the parameters. This property allows us to replace the inner product with a feature map $ \phi(x_i) $, affecting only the inner product terms. This modularityseparating the classification method (SVM) from the feature representation $ \phi(x) $provides flexibility to explore these two aspects independently.

In this section, we explore the feature representation $ \phi(x) $ and introduce the concept of **kernels**, though we do not delve into technical details.

### Kernels and Non-Linear Classification

Since $ \phi(x) $ can be a non-linear function, we can use an SVM (which assumes a linear classifier in the feature space) to construct classifiers that are non-linear in the original input space of the examples $ x_n $. This approach provides a second method, alongside the soft margin SVM, to handle datasets that are not linearly separable in the original space.

Many algorithms and statistical methods share this property observed in the dual SVM: the only inner products are between examples. Instead of explicitly defining a non-linear feature map $ \phi(\cdot) $ and computing the inner product $ \langle \phi(x_i), \phi(x_j) \rangle $, we define a **similarity function** $ k(x_i, x_j) $ between examples $ x_i $ and $ x_j $. For a specific class of similarity functions called **kernels**, this function implicitly defines a non-linear feature map $ \phi(\cdot) $.

#### Definition of a Kernel

Kernels are functions $ k: X \times X \to \mathbb{R} $ for which there exists a Hilbert space $ H $ and a feature map $ \phi: X \to H $ such that:

$$
k(x_i, x_j) = \langle \phi(x_i), \phi(x_j) \rangle_H. \tag{12.52}
$$

There is a unique **reproducing kernel Hilbert space (RKHS)** associated with every kernel $ k $ (Aronszajn, 1950; Berlinet and Thomas-Agnan, 2004). In this unique association, the **canonical feature map** is given by $ \phi(x) = k(\cdot, x) $.

The generalization from an inner product to a kernel function (Equation 12.52) is known as the **kernel trick** (Schlkopf and Smola, 2002; Shawe-Taylor and Cristianini, 2004). The kernel trick allows us to work with the kernel function directly, hiding the explicit computation of the non-linear feature map $ \phi(\cdot) $.

### Kernel Matrix (Gram Matrix)

The matrix $ K \in \mathbb{R}^{N \times N} $, constructed by applying the kernel function $ k(\cdot, \cdot) $ to all pairs of examples in a dataset (or computing their inner products in the feature space), is called the **Gram matrix** or **kernel matrix**. Kernels must be symmetric and positive semidefinite to ensure that the kernel matrix $ K $ satisfies:

$$
\forall z \in \mathbb{R}^N: z^\top K z \geq 0. \tag{12.53}
$$

This property ensures that $ K $ is a valid kernel matrix, as discussed in Section 3.2.3.

### Examples of Kernels

Some popular kernels for multivariate real-valued data $ x_i \in \mathbb{R}^D $ include:
- **Polynomial kernel**
- **Gaussian radial basis function (RBF) kernel**
- **Rational quadratic kernel**

These kernels are widely used in practice (Schlkopf and Smola, 2002; Rasmussen and Williams, 2006). Figure 12.10 illustrates the effect of different kernels on the decision boundaries of an SVM on an example dataset:
- **Figure 12.10(a)**: SVM with a linear kernel, resulting in a linear decision boundary.
- **Figure 12.10(b)**: SVM with an RBF kernel, producing a non-linear decision boundary.
- **Figure 12.10(c)**: SVM with a polynomial kernel (degree 2), showing a non-linear boundary.
- **Figure 12.10(d)**: SVM with a polynomial kernel (degree 3), showing a more complex non-linear boundary.

Note that while the decision boundaries appear non-linear in the original space, the underlying problem in the feature space is still a linear separating hyperplane, transformed by the kernel function.

### Remark: Multiple Meanings of "Kernel"

The term "kernel" has multiple meanings in machine learning, which can be confusing:
- In this chapter, "kernel" refers to the reproducing kernel Hilbert space (RKHS) context (Aronszajn, 1950; Saitoh, 1988).
- In linear algebra (Section 2.7.3), the "kernel" refers to the null space of a matrix.
- In kernel density estimation (Section 11.5), a "kernel" is a smoothing function used to estimate probability densities.

$ \diamond $

### Computational Efficiency of Kernels

Since the explicit feature map $ \phi(x) $ is mathematically equivalent to the kernel function $ k(x_i, x_j) $, practitioners often design kernel functions that are more efficient to compute than the inner product of explicit feature maps. For example:
- **Polynomial Kernel**: The explicit feature map for a polynomial kernel of degree $ d $ in $ D $ dimensions results in a feature space with a number of terms that grows rapidly (e.g., $ \binom{D+d}{d} $ terms). However, the kernel function $ k(x_i, x_j) = (x_i^\top x_j + c)^d $ requires only one multiplication per input dimension, offering significant computational savings (Schlkopf and Smola, 2002).
- **Gaussian RBF Kernel**: The corresponding feature space is infinite-dimensional, making explicit computation of $ \phi(x) $ infeasible. However, the kernel $ k(x_i, x_j) = \exp\left(-\frac{\|x_i - x_j\|^2}{2\sigma^2}\right) $ can be computed directly, allowing us to work with infinite-dimensional spaces implicitly (Schlkopf and Smola, 2002; Rasmussen and Williams, 2006).

### Flexibility of Kernels

Another advantage of the kernel trick is its flexibility with input data types. The inner product in Equation (12.52) is defined on the output of $ \phi(\cdot) $, but the input space $ X $ is not restricted to real numbers. Thus, kernels can be defined on various objects, such as sets, sequences, strings, graphs, and distributions (Ben-Hur et al., 2008; Grtner, 2008; Shi et al., 2009; Sriperumbudur et al., 2010; Vishwanathan et al., 2010). This allows SVMs to be applied to a wide range of data types beyond multivariate real-valued data.

````python
import math
import matplotlib.pyplot as plt

class KernelSVM:
    """
    Soft Margin SVM with kernels in 2D using core Python.
    """
    
    def __init__(self, kernel='linear', C=1.0, degree=2, sigma=1.0):
        """
        Initialize the SVM with a specified kernel.
        
        Parameters:
        -----------
        kernel : str
            Type of kernel ('linear', 'polynomial', 'rbf')
        C : float
            Soft margin penalty parameter
        degree : int
            Degree for polynomial kernel
        sigma : float
            Parameter for RBF kernel
        """
        self.kernel_type = kernel
        self.C = C
        self.degree = degree
        self.sigma = sigma
        self.alpha = None
        self.support_vectors = None
        self.b = None
    
    def dot_product(self, x1, x2):
        """Compute dot product of two 2D vectors."""
        return x1[0] * x2[0] + x1[1] * x2[1]
    
    def norm(self, x):
        """Compute the Euclidean norm of a 2D vector."""
        return math.sqrt(self.dot_product(x, x))
    
    def kernel(self, xi, xj):
        """
        Compute the kernel function k(xi, xj).
        
        Parameters:
        -----------
        xi, xj : tuple
            2D points (x1, x2)
            
        Returns:
        --------
        float : Kernel value
        """
        if self.kernel_type == 'linear':
            return self.dot_product(xi, xj)
        elif self.kernel_type == 'polynomial':
            return (self.dot_product(xi, xj) + 1) ** self.degree
        elif self.kernel_type == 'rbf':
            diff = (xi[0] - xj[0], xi[1] - xj[1])
            return math.exp(-self.norm(diff) ** 2 / (2 * self.sigma ** 2))
        else:
            raise ValueError("Unsupported kernel type")
    
    def fit(self, X, y):
        """
        Fit the SVM using a simplified approach to approximate the dual solution.
        
        Parameters:
        -----------
        X : list of tuples
            2D data points [(x1, x2), ...]
        y : list of int
            Labels (+1 or -1)
        """
        N = len(X)
        # Simplified: Manually identify support vectors and set alpha values
        # In a real implementation, solve the dual optimization problem:
        # Maximize sum(alpha_n) - 0.5 * sum_i sum_j alpha_i alpha_j y_i y_j k(x_i, x_j)
        # Subject to sum(alpha_n y_n) = 0, 0 <= alpha_n <= C
        
        # For this dataset, we approximate by selecting boundary points as support vectors
        # and setting alpha values to balance the constraints
        pos_points = [(X[i], i) for i in range(N) if y[i] == 1]
        neg_points = [(X[i], i) for i in range(N) if y[i] == -1]
        
        # Approximate support vectors: points closest to the opposite class
        sv_pos, sv_neg = None, None
        min_dist = float('inf')
        for p1, idx1 in pos_points:
            for p2, idx2 in neg_points:
                dist = self.norm((p1[0] - p2[0], p1[1] - p2[1]))
                if dist < min_dist:
                    min_dist = dist
                    sv_pos = (p1, idx1)
                    sv_neg = (p2, idx2)
        
        # Set alpha values (simplified: assume these are the only support vectors)
        self.support_vectors = [(sv_pos[0], 1, sv_pos[1]), (sv_neg[0], -1, sv_neg[1])]
        # alpha values should satisfy sum(alpha_n y_n) = 0
        # Let alpha_1 = alpha_2 for simplicity, and within bounds [0, C]
        alpha_1 = min(self.C, 1.0)  # Arbitrary choice within bounds
        self.alpha = [0.0] * N
        self.alpha[sv_pos[1]] = alpha_1
        self.alpha[sv_neg[1]] = alpha_1  # Since y_1 = +1, y_2 = -1, satisfies constraint
        
        # Compute bias b using a support vector
        # For a support vector: y_n (sum(alpha_m y_m k(x_m, x_n)) + b) = 1
        sv_x, sv_y, sv_idx = self.support_vectors[0]
        sum_term = 0.0
        for m in range(N):
            if self.alpha[m] > 0:
                sum_term += self.alpha[m] * y[m] * self.kernel(X[m], sv_x)
        self.b = sv_y - sum_term
    
    def predict(self, x, X, y):
        """
        Predict the class of a 2D point.
        
        Parameters:
        -----------
        x : tuple
            2D point (x1, x2)
        X : list of tuples
            Training data
        y : list of int
            Training labels
            
        Returns:
        --------
        int : Predicted class (+1 or -1)
        """
        sum_term = 0.0
        for n in range(len(X)):
            if self.alpha[n] > 0:
                sum_term += self.alpha[n] * y[n] * self.kernel(X[n], x)
        return 1 if sum_term + self.b > 0 else -1

def generate_half_circle_data():
    """Generate a dataset of two interleaving half-circles."""
    X = []
    y = []
    # Class 1: Upper half-circle
    for theta in [i * math.pi / 10 for i in range(11)]:
        x1 = 2 * math.cos(theta)
        x2 = 2 * math.sin(theta) + 1
        X.append((x1, x2))
        y.append(1)
    # Class 2: Lower half-circle
    for theta in [i * math.pi / 10 for i in range(11)]:
        x1 = 1.5 * math.cos(theta + math.pi)
        x2 = 1.5 * math.sin(theta + math.pi) - 1
        X.append((x1, x2))
        y.append(-1)
    return X, y

def plot_svm_decision_boundary(X, y, svm, kernel_name, ax):
    """Plot the decision boundary and data points."""
    # Plot data points
    pos_points = [X[i] for i in range(len(X)) if y[i] == 1]
    neg_points = [X[i] for i in range(len(X)) if y[i] == -1]
    ax.scatter([p[0] for p in pos_points], [p[1] for p in pos_points], color='blue', label='Class +1', s=100)
    ax.scatter([p[0] for p in neg_points], [p[1] for p in neg_points], color='orange', label='Class -1', s=100)
    
    # Plot support vectors
    for sv_x, sv_y, _ in svm.support_vectors:
        color = 'blue' if sv_y == 1 else 'orange'
        ax.scatter([sv_x[0]], [sv_x[1]], color=color, marker='*', s=200, edgecolors='black')
    
    # Create a grid to plot the decision boundary
    x1_min, x1_max = min([p[0] for p in X]) - 1, max([p[0] for p in X]) + 1
    x2_min, x2_max = min([p[1] for p in X]) - 1, max([p[1] for p in X]) + 1
    grid_points = []
    for x1 in [x1_min + i * (x1_max - x1_min) / 50 for i in range(51)]:
        for x2 in [x2_min + i * (x2_max - x2_min) / 50 for i in range(51)]:
            grid_points.append((x1, x2))
    
    # Predict on the grid
    predictions = [svm.predict(p, X, y) for p in grid_points]
    grid_x1 = [p[0] for p in grid_points]
    grid_x2 = [p[1] for p in grid_points]
    
    # Plot decision boundary (contour where prediction changes)
    ax.scatter(grid_x1, grid_x2, c=['blue' if pred == 1 else 'orange' for pred in predictions], alpha=0.1, s=10)
    ax.set_title(f'SVM with {kernel_name} Kernel')
    ax.set_xlabel('First Feature')
    ax.set_ylabel('Second Feature')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

if __name__ == "__main__":
    print("Kernel SVM Implementation in Core Python")
    print("=" * 60)
    
    # Generate dataset
    X, y = generate_half_circle_data()
    
    # Create subplots for different kernels (mimicking Figure 12.10)
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.flatten()
    
    # (a) Linear Kernel
    svm_linear = KernelSVM(kernel='linear', C=1.0)
    svm_linear.fit(X, y)
    plot_svm_decision_boundary(X, y, svm_linear, 'Linear', axes[0])
    
    # (b) RBF Kernel
    svm_rbf = KernelSVM(kernel='rbf', C=1.0, sigma=1.0)
    svm_rbf.fit(X, y)
    plot_svm_decision_boundary(X, y, svm_rbf, 'RBF', axes[1])
    
    # (c) Polynomial Kernel (degree 2)
    svm_poly2 = KernelSVM(kernel='polynomial', C=1.0, degree=2)
    svm_poly2.fit(X, y)
    plot_svm_decision_boundary(X, y, svm_poly2, 'Polynomial (degree 2)', axes[2])
    
    # (d) Polynomial Kernel (degree 3)
    svm_poly3 = KernelSVM(kernel='polynomial', C=1.0, degree=3)
    svm_poly3.fit(X, y)
    plot_svm_decision_boundary(X, y, svm_poly3, 'Polynomial (degree 3)', axes[3])
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== Demonstration completed ===")
````
