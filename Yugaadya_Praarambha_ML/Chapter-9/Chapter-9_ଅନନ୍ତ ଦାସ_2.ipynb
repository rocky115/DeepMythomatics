## Finding Optimal Coordinates

We now focus on finding the optimal coordinates $z_{1n}, \ldots, z_{Mn}$ of the projections $\tilde{x}_n$ for $n = 1, \ldots, N$. Consider Figure 10.7(b), where the principal subspace $U$ is spanned by a single vector $b$. Geometrically, finding the optimal coordinates $z$ means identifying the linear projection $\tilde{x}$ with respect to $b$ that minimizes the distance $\lVert \tilde{x} - x \rVert$. From Figure 10.7(b), it’s evident that this corresponds to the orthogonal projection, and we will confirm this analytically.

Assume an orthonormal basis (ONB) $(b_1, \ldots, b_M)$ of the subspace $U \subseteq \mathbb{R}^D$ with $\text{dim}(U) = M$. To find the optimal coordinates $z_{in}$ with respect to this basis, we compute the partial derivatives of the reconstruction error $J_M$ (Equation 10.29):

$$
\frac{\partial J_M}{\partial z_{in}} = \frac{\partial J_M}{\partial \tilde{x}_n} \frac{\partial \tilde{x}_n}{\partial z_{in}}, \tag{10.30a}
$$

$$
\frac{\partial J_M}{\partial \tilde{x}_n} = -\frac{2}{N} (x_n - \tilde{x}_n)^\top \in \mathbb{R}^{1 \times D}, \tag{10.30b}
$$

where $J_M = \frac{1}{N} \sum_{n=1}^N \lVert x_n - \tilde{x}_n \rVert^2$. Next, we compute the derivative of the reconstruction $\tilde{x}_n$ (Equation 10.28) with respect to $z_{in}$:

$$
\frac{\partial \tilde{x}_n}{\partial z_{in}} = \frac{\partial}{\partial z_{in}} \left( \sum_{m=1}^M z_{mn} b_m \right) = b_i, \tag{10.30c}
$$

for $i = 1, \ldots, M$, since the derivative with respect to $z_{in}$ selects the $i$-th basis vector $b_i$.

Substituting Equation (10.30c) into (10.30b), we get:

$$
\frac{\partial J_M}{\partial z_{in}} = -\frac{2}{N} (x_n - \tilde{x}_n)^\top b_i = -\frac{2}{N} \left( x_n - \sum_{m=1}^M z_{mn} b_m \right)^\top b_i, \tag{10.31a}
$$

$$
= -\frac{2}{N} (x_n^\top b_i - z_{in} b_i^\top b_i) = -\frac{2}{N} (x_n^\top b_i - z_{in}), \tag{10.31b}
$$

since $b_i^\top b_i = 1$ (because $(b_1, \ldots, b_M)$ is an ONB). Setting this partial derivative to 0:

$$
-\frac{2}{N} (x_n^\top b_i - z_{in}) = 0,
$$

yields the optimal coordinates:

$$
z_{in} = x_n^\top b_i = b_i^\top x_n, \tag{10.32}
$$

for $i = 1, \ldots, M$ and $n = 1, \ldots, N$. This result shows that the optimal coordinates $z_{in}$ of the projection $\tilde{x}_n$ are the coordinates of the orthogonal projection (Section 3.8) of the original data point $x_n$ onto the one-dimensional subspace spanned by $b_i$. Consequently, the optimal linear projection $\tilde{x}_n$ of $x_n$ is an orthogonal projection.

The coordinates of $\tilde{x}_n$ with respect to the basis $(b_1, \ldots, b_M)$ are the coordinates of the orthogonal projection of $x_n$ onto the principal subspace $U$. An orthogonal projection is the best linear mapping for the objective of minimizing the reconstruction error (Equation 10.29). Furthermore, the coordinates $\zeta_m$ of $x$ in Equation (10.26) and $z_m$ of $\tilde{x}$ in Equation (10.27) must be identical for $m = 1, \ldots, M$, since $U^\perp = \text{span}[b_{M+1}, \ldots, b_D]$ is the orthogonal complement (Section 3.6) of $U = \text{span}[b_1, \ldots, b_M]$.

### Remark: Orthogonal Projections with Orthonormal Basis Vectors

Let’s recap orthogonal projections from Section 3.8. If $(b_1, \ldots, b_D)$ is an orthonormal basis of $\mathbb{R}^D$, the orthogonal projection of $x$ onto the subspace spanned by the $j$-th basis vector $b_j$ is:

$$
\tilde{x} = b_j (b_j^\top b_j)^{-1} b_j^\top x = b_j b_j^\top x \in \mathbb{R}^D, \tag{10.33}
$$

since $b_j^\top b_j = 1$. The coordinate of this projection with respect to $b_j$ is $z_j = b_j^\top x$, so $z_j b_j = \tilde{x}$. Figure 10.8(b) illustrates this setting, showing the orthogonal projection of a vector $x \in \mathbb{R}^2$ onto a one-dimensional subspace $U$, with the coordinate $z_1$ scaling $b$ to reach $\tilde{x}$.

More generally, for an $M$-dimensional subspace of $\mathbb{R}^D$ with orthonormal basis vectors $b_1, \ldots, b_M$, the orthogonal projection of $x$ onto the subspace is:

$$
\tilde{x} = B (B^\top B)^{-1} B^\top x = B B^\top x, \tag{10.34}
$$

where $B := [b_1, \ldots, b_M] \in \mathbb{R}^{D \times M}$, and $B^\top B = I$ (since the columns of $B$ are orthonormal). The coordinates of this projection with respect to the basis $(b_1, \ldots, b_M)$ are $z := B^\top x$, as discussed in Section 3.8. These coordinates represent the projected vector in the new coordinate system defined by $(b_1, \ldots, b_M)$. Although $\tilde{x} \in \mathbb{R}^D$, we only need $M$ coordinates $z_1, \ldots, z_M$ to represent it; the other $D-M$ coordinates with respect to the basis vectors $(b_{M+1}, \ldots, b_D)$ are always 0. $\diamond$

### Summary

So far, we have shown that for a given ONB, the optimal coordinates of $\tilde{x}_n$ are obtained by an orthogonal projection onto the principal subspace. The coordinates of the optimal projection of $x_n$ with respect to the basis vectors $b_1, \ldots, b_M$ are the coordinates of the orthogonal projection of $x_n$ onto the principal subspace. In the next section, we will determine the optimal basis for this subspace.

```python
import math
import matplotlib.pyplot as plt

# --- Define the Dataset ---
# 2D dataset (D = 2) with N = 5 points (same as previous for continuity)
data = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N = len(data)
D = len(data[0])
M = 1  # Project onto a 1D subspace

# --- Center the Data ---
def compute_mean(data):
    """Compute the mean of the data across each dimension."""
    mean = [0.0] * D
    for x in data:
        for i in range(D):
            mean[i] += x[i] / N
    return mean

def center_data(data, mean):
    """Center the data by subtracting the mean."""
    centered_data = []
    for x in data:
        centered_x = [x[i] - mean[i] for i in range(D)]
        centered_data.append(centered_x)
    return centered_data

mean = compute_mean(data)
X = center_data(data, mean)
print("Original Data:", data)
print("Mean:", mean)
print("Centered Data (X):", X)

# --- Define the Basis ---
# For simplicity, use the first principal component as b1 (as computed previously)
# Alternatively, we could use any unit vector; here we use the optimal b1 for demonstration
b1 = [0.906, 0.423]  # From prior PCA computation (eigenvector of S with largest eigenvalue)
# Normalize to ensure it's a unit vector (though it already is)
norm = math.sqrt(sum(bi**2 for bi in b1))
b1 = [bi / norm for bi in b1]
B = [b1]  # Basis for the M=1 dimensional subspace
print(f"\nBasis Vector b1: {b1}")

# --- Step 1: Compute Optimal Coordinates z_in (Equation 10.32) ---
def compute_coordinates(X, B):
    """Compute z_in = b_i^T x_n for each data point (Equation 10.32)."""
    Z = []
    for x in X:
        z_n = []
        for b in B:
            # z_in = b_i^T x_n
            z_in = sum(b[i] * x[i] for i in range(D))
            z_n.append(z_in)
        Z.append(z_n)
    return Z

Z = compute_coordinates(X, B)
print("\nOptimal Coordinates z_in:")
for i, z in enumerate(Z, 1):
    print(f"z_{i}: {z}")

# --- Compute Reconstructions x̃_n (Equation 10.28) ---
def compute_reconstruction(Z, B):
    """Compute x̃_n = B z_n (Equation 10.28)."""
    X_tilde = []
    for z in Z:
        x_tilde = [0.0] * D
        for m in range(len(B)):
            for i in range(D):
                x_tilde[i] += B[m][i] * z[m]
        X_tilde.append(x_tilde)
    return X_tilde

X_tilde = compute_reconstruction(Z, B)
print("\nReconstructed Data (x̃_n):")
for x_tilde in X_tilde:
    print(x_tilde)

# --- Compute Reconstruction Error (Equation 10.29) ---
def compute_reconstruction_error(X, X_tilde):
    """Compute J_M = (1/N) Σ ||x_n - x̃_n||^2 (Equation 10.29)."""
    total_error = 0.0
    for n in range(N):
        squared_norm = sum((X[n][i] - X_tilde[n][i])**2 for i in range(D))
        total_error += squared_norm
    J_M = total_error / N
    return J_M

J_M = compute_reconstruction_error(X, X_tilde)
print(f"\nReconstruction Error J_{M}: {J_M:.3f}")

# --- Verify Orthogonal Projection ---
def verify_orthogonal_projection(X, X_tilde, B):
    """Verify that (x_n - x̃_n) is orthogonal to b_i for all i."""
    print("\nVerifying Orthogonal Projection:")
    for i, b in enumerate(B, 1):
        print(f"\nChecking orthogonality with b{i}: {b}")
        for n in range(N):
            # Compute x_n - x̃_n
            diff = [X[n][j] - X_tilde[n][j] for j in range(D)]
            # Dot product (x_n - x̃_n)^T b_i should be 0
            dot_product = sum(diff[j] * b[j] for j in range(D))
            print(f"Dot product (x_{n+1} - x̃_{n+1})^T b{i}: {dot_product:.3e}")

# --- Main Demonstration ---
def demonstrate_pca_coordinates():
    """Demonstrate finding optimal coordinates in PCA (Section 10.3.2)."""
    print("=== PCA Optimal Coordinates Demonstration ===")
    print("Section 10.3.2: Finding Optimal Coordinates\n")
    
    print(f"Optimal Coordinates z_in for M={M}-dimensional Subspace:")
    for i, z in enumerate(Z, 1):
        print(f"z_{i}: {z}")
    print(f"Reconstruction Error J_{M}: {J_M:.3f}")
    
    # Verify orthogonality
    verify_orthogonal_projection(X, X_tilde, B)

# --- Plotting Function ---
def plot_pca_coordinates(X, X_tilde, B):
    """Plot the centered data, reconstructed data, principal subspace, and difference vectors."""
    plt.figure(figsize=(8, 6))
    
    # Plot centered data
    cent_x = [x[0] for x in X]
    cent_y = [x[1] for x in X]
    plt.scatter(cent_x, cent_y, color='green', label='Centered Data', alpha=0.7)
    
    # Plot reconstructed data
    recon_x = [x[0] for x in X_tilde]
    recon_y = [x[1] for x in X_tilde]
    plt.scatter(recon_x, recon_y, color='orange', label='Reconstructed Data (x̃_n)', alpha=0.7)
    
    # Plot difference vectors (x_n - x̃_n)
    for n in range(N):
        plt.plot([X[n][0], X_tilde[n][0]], [X[n][1], X_tilde[n][1]], 'r--', alpha=0.5,
                 label='Difference (x_n - x̃_n)' if n == 0 else "")
    
    # Plot the principal subspace (line spanned by b1)
    scale = 3  # Scale for visibility
    b1 = B[0]
    plt.plot([-scale * b1[0], scale * b1[0]], [-scale * b1[1], scale * b1[1]],
             color='blue', label='Principal Subspace (U)')
    
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('PCA: Finding Optimal Coordinates (Orthogonal Projection)')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PCA Optimal Coordinates Demonstration with Matplotlib Plotting")
    print("=" * 60)

    # Demonstrate PCA
    demonstrate_pca_coordinates()

    # Plot the results
    print("\nPlotting PCA Results")
    plot_pca_coordinates(X, X_tilde, B)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Basis Vector b1: {b1}")
    print(f"• Reconstruction Error J_{M}: {J_M:.3f}")
    print("• Plotted centered data, reconstructed data, difference vectors, and principal subspace")
```

## Finding the Basis of the Principal Subspace

To determine the basis vectors $b_1, \ldots, b_M$ of the principal subspace, we reformulate the loss function $J_M$ (Equation 10.29) using the results from Section 10.3.2. This reformulation simplifies the process of finding the optimal basis vectors.

### Reformulating the Loss Function

Using the optimal coordinates from Equation (10.32), the projection $\tilde{x}_n$ is:

$$
\tilde{x}_n = \sum_{m=1}^M z_{mn} b_m = \sum_{m=1}^M (x_n^\top b_m) b_m, \tag{10.35}
$$

Exploiting the symmetry of the dot product (\( x_n^\top b_m = b_m^\top x_n \)), we rewrite $\tilde{x}_n$ as:

$$
\tilde{x}_n = \sum_{m=1}^M b_m (b_m^\top x_n). \tag{10.36}
$$

This can be expressed in matrix form as $\tilde{x}_n = B B^\top x_n$, where $B = [b_1, \ldots, b_M] \in \mathbb{R}^{D \times M}$ is the matrix whose columns are the basis vectors.

Next, consider the original data point $x_n$. Since $(b_1, \ldots, b_D)$ is an orthonormal basis of $\mathbb{R}^D$, we can write $x_n$ as a linear combination of all basis vectors:

$$
x_n = \sum_{d=1}^D z_{dn} b_d = \sum_{d=1}^D (x_n^\top b_d) b_d = \sum_{d=1}^D b_d (b_d^\top x_n), \tag{10.37a}
$$

Splitting the sum into contributions from the principal subspace (spanned by $b_1, \ldots, b_M$) and its orthogonal complement (spanned by $b_{M+1}, \ldots, b_D$):

$$
x_n = \sum_{m=1}^M b_m (b_m^\top x_n) + \sum_{j=M+1}^D b_j (b_j^\top x_n), \tag{10.37b}
$$

Subtracting $\tilde{x}_n$ (Equation 10.36) from $x_n$, the displacement vector $x_n - \tilde{x}_n$ is:

$$
x_n - \tilde{x}_n = \left( \sum_{m=1}^M b_m (b_m^\top x_n) + \sum_{j=M+1}^D b_j (b_j^\top x_n) \right) - \sum_{m=1}^M b_m (b_m^\top x_n) = \sum_{j=M+1}^D b_j (b_j^\top x_n), \tag{10.38a}
$$

$$
x_n - \tilde{x}_n = \sum_{j=M+1}^D (x_n^\top b_j) b_j. \tag{10.38b}
$$

This shows that the displacement vector $x_n - \tilde{x}_n$ is the projection of $x_n$ onto the orthogonal complement of the principal subspace $U$, i.e., the subspace $U^\perp = \text{span}[b_{M+1}, \ldots, b_D]$. The matrix $\sum_{j=M+1}^D b_j b_j^\top$ in Equation (10.38a) is the projection matrix onto $U^\perp$. Thus, the displacement vector lies entirely in the subspace orthogonal to the principal subspace, as illustrated in Figure 10.9, which shows data points $x_n$ (blue), their projections $\tilde{x}_n$ (orange), and the displacement vectors $x_n - \tilde{x}_n$ lying in $U^\perp$.

### Remark: Low-Rank Approximation

From Equation (10.36), the projection matrix that maps $x_n$ to $\tilde{x}_n$ is:

$$
\sum_{m=1}^M b_m b_m^\top = B B^\top, \tag{10.39}
$$

where $B = [b_1, \ldots, b_M]$. Since each $b_m b_m^\top$ is a rank-one matrix, $B B^\top$ is a sum of $M$ rank-one matrices, making it a rank-$M$ matrix. This projection can be viewed as a low-rank approximation of the data in the principal subspace.

### Reformulating the Reconstruction Error (Continued)

The average squared reconstruction error from Equation (10.29) can be expressed using the projection matrix $ B B^\top $:

$$
\frac{1}{N} \sum_{n=1}^N \lVert x_n - \tilde{x}_n \rVert^2 = \frac{1}{N} \sum_{n=1}^N \lVert x_n - B B^\top x_n \rVert^2, \tag{10.40a}
$$

$$
= \frac{1}{N} \sum_{n=1}^N \lVert (I - B B^\top) x_n \rVert^2, \tag{10.40b}
$$

where $ B = [b_1, \ldots, b_M] \in \mathbb{R}^{D \times M} $, and $ I - B B^\top $ is the projection matrix onto the orthogonal complement $ U^\perp $. Finding the orthonormal basis vectors $ b_1, \ldots, b_M $ that minimize the difference between the original data $ x_n $ and their projections $ \tilde{x}_n $ is equivalent to finding the best rank-$ M $ approximation $ B B^\top $ of the identity matrix $ I $ (see Section 4.6). This completes the low-rank approximation remark. $\diamond$

### Deriving the Optimal Basis

Using the displacement vector from Equation (10.38b), the reconstruction error \( J_M \) is:

$$
J_M = \frac{1}{N} \sum_{n=1}^N \lVert x_n - \tilde{x}_n \rVert^2 = \frac{1}{N} \sum_{n=1}^N \left\lVert \sum_{j=M+1}^D (b_j^\top x_n) b_j \right\rVert^2, \tag{10.41}
$$

Expanding the squared norm and using the fact that $ b_j $ form an orthonormal basis $ b_j^\top b_k = 0 $ for $ j \neq k $, $ b_j^\top b_j = 1 $):

$$
J_M = \frac{1}{N} \sum_{n=1}^N \sum_{j=M+1}^D (b_j^\top x_n)^2 = \frac{1}{N} \sum_{n=1}^N \sum_{j=M+1}^D b_j^\top x_n x_n^\top b_j, \tag{10.42a}
$$

Using the symmetry of the dot product $( b_j^\top x_n = x_n^\top b_j $):

$$
J_M = \frac{1}{N} \sum_{n=1}^N \sum_{j=M+1}^D b_j^\top x_n x_n^\top b_j, \tag{10.42b}
$$

Swapping the order of summation:

$$
J_M = \sum_{j=M+1}^D b_j^\top \left( \frac{1}{N} \sum_{n=1}^N x_n x_n^\top \right) b_j = \sum_{j=M+1}^D b_j^\top S b_j, \tag{10.43a}
$$

where $ S := \frac{1}{N} \sum_{n=1}^N x_n x_n^\top $ is the data covariance matrix (since the dataset is centered, i.e., $ \mathbb{E}[X] = 0 $).

Using the trace operator (Section 4.18), which is linear and invariant to cyclic permutations:

$$
J_M = \sum_{j=M+1}^D \text{tr}(b_j^\top S b_j) = \sum_{j=M+1}^D \text{tr}(S b_j b_j^\top) = \text{tr} \left( \left( \sum_{j=M+1}^D b_j b_j^\top \right) S \right), \tag{10.43b}
$$

where $ \sum_{j=M+1}^D b_j b_j^\top $ is the projection matrix onto the orthogonal complement $ U^\perp $, which has rank $ D - M $.

Equation (10.43a) shows that the average squared reconstruction error is the variance of the data projected onto the orthogonal complement of the principal subspace. Thus, minimizing the reconstruction error is equivalent to minimizing the variance in the subspace we ignore (i.e., $ U^\perp $). Equivalently, this means maximizing the variance retained in the principal subspace $ U $, linking the projection perspective to the maximum-variance formulation of PCA in Section 10.2. Since the solution is the same as in the maximum-variance perspective, we omit a redundant derivation and summarize the results in the context of the projection perspective.

### Optimal Basis and Reconstruction Error

The average squared reconstruction error, when projecting onto the \( M \)-dimensional principal subspace, is:

$$
J_M = \sum_{j=M+1}^D \lambda_j, \tag{10.44}
$$

where $ \lambda_j $ are the eigenvalues of the data covariance matrix $ S $. To minimize $ J_M $, we must select the smallest $ D - M $ eigenvalues, meaning the corresponding eigenvectors $ b_{M+1}, \ldots, b_D $ form the basis of the orthogonal complement $ U^\perp $. Consequently, the basis of the principal subspace $ U $ consists of the eigenvectors $ b_1, \ldots, b_M $ associated with the largest $ M $ eigenvalues of $ S $.

Minimizing the average squared reconstruction error is thus equivalent to maximizing the variance of the projected data in the principal subspace, reinforcing the connection between the projection and maximum-variance perspectives of PCA.

Notably, the variation within the set of "0"s is significantly greater than within the set of "1"s, reflecting differences in the complexity of the digit patterns.

```python
import math
import matplotlib.pyplot as plt

# --- Define the Dataset ---
# 2D dataset (D = 2) with N = 5 points (same as previous for continuity)
data = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N = len(data)
D = len(data[0])
M = 1  # Project onto a 1D subspace

# --- Center the Data ---
def compute_mean(data):
    """Compute the mean of the data across each dimension."""
    mean = [0.0] * D
    for x in data:
        for i in range(D):
            mean[i] += x[i] / N
    return mean

def center_data(data, mean):
    """Center the data by subtracting the mean."""
    centered_data = []
    for x in data:
        centered_x = [x[i] - mean[i] for i in range(D)]
        centered_data.append(centered_x)
    return centered_data

mean = compute_mean(data)
X = center_data(data, mean)
print("Original Data:", data)
print("Mean:", mean)
print("Centered Data (X):", X)

# --- Compute the Covariance Matrix S ---
def compute_covariance_matrix(data):
    """Compute the data covariance matrix S = (1/N) X X^T."""
    S = [[0.0 for _ in range(D)] for _ in range(D)]
    for x in data:
        for i in range(D):
            for j in range(D):
                S[i][j] += (x[i] * x[j]) / N
    return S

S = compute_covariance_matrix(X)
print("\nCovariance Matrix S:")
for row in S:
    print(row)

# --- Eigenvalue Decomposition (Manual for 2x2 Matrix) ---
def compute_eigenvalues_2x2(S):
    """Compute eigenvalues of a 2x2 matrix using the characteristic equation."""
    a, b = S[0][0], S[0][1]
    c, d = S[1][0], S[1][1]
    trace = a + d
    det = a * d - b * c
    discriminant = trace**2 - 4 * det
    if discriminant < 0:
        raise ValueError("Complex eigenvalues not handled.")
    sqrt_disc = math.sqrt(discriminant)
    lambda1 = (trace + sqrt_disc) / 2
    lambda2 = (trace - sqrt_disc) / 2
    return lambda1, lambda2  # Returns in descending order

def compute_eigenvector_2x2(S, eigenvalue):
    """Compute the eigenvector for a given eigenvalue (normalized)."""
    a, b = S[0][0], S[0][1]
    c, d = S[1][0], S[1][1]
    if abs(a - eigenvalue) > 1e-10:
        v1 = -b / (a - eigenvalue)
        v2 = 1.0
    else:
        v1 = 1.0
        v2 = -(d - eigenvalue) / c if abs(c) > 1e-10 else 0.0
    norm = math.sqrt(v1**2 + v2**2)
    return [v1 / norm, v2 / norm]

# Compute eigenvalues and eigenvectors
eigenvalues = compute_eigenvalues_2x2(S)
b1 = compute_eigenvector_2x2(S, eigenvalues[0])
b2 = compute_eigenvector_2x2(S, eigenvalues[1])
basis_vectors = [b1, b2]
B = basis_vectors[:M]  # Basis for the M=1 dimensional subspace
print(f"\nEigenvalues: λ1 = {eigenvalues[0]:.3f}, λ2 = {eigenvalues[1]:.3f}")
print(f"Principal Subspace Basis b1: {b1}")
print(f"Orthogonal Complement Basis b2: {b2}")

# --- Compute Projections x̃_n = B B^T x_n (Equation 10.36) ---
def compute_projection_matrix(B):
    """Compute B B^T = Σ b_m b_m^T."""
    BBT = [[0.0 for _ in range(D)] for _ in range(D)]
    for b in B:
        for i in range(D):
            for j in range(D):
                BBT[i][j] += b[i] * b[j]
    return BBT

def matrix_vector_multiply(matrix, vector):
    """Multiply a DxD matrix by a D-dimensional vector."""
    result = [0.0] * D
    for i in range(D):
        for j in range(D):
            result[i] += matrix[i][j] * vector[j]
    return result

BBT = compute_projection_matrix(B)
X_tilde = [matrix_vector_multiply(BBT, x) for x in X]
print("\nReconstructed Data (x̃_n):")
for x_tilde in X_tilde:
    print(x_tilde)

# --- Compute Displacement Vectors x_n - x̃_n (Equation 10.38) ---
displacements = []
for n in range(N):
    disp = [X[n][i] - X_tilde[n][i] for i in range(D)]
    displacements.append(disp)
print("\nDisplacement Vectors (x_n - x̃_n):")
for disp in displacements:
    print(disp)

# --- Compute Reconstruction Error J_M (Equations 10.41–10.44) ---
def compute_reconstruction_error(X, X_tilde):
    """Compute J_M = (1/N) Σ ||x_n - x̃_n||^2 (Equation 10.41)."""
    total_error = 0.0
    for n in range(N):
        squared_norm = sum((X[n][i] - X_tilde[n][i])**2 for i in range(D))
        total_error += squared_norm
    J_M = total_error / N
    return J_M

J_M = compute_reconstruction_error(X, X_tilde)
print(f"\nReconstruction Error J_{M}: {J_M:.3f}")

# --- Verify J_M = Σ λ_j for j=M+1 to D (Equation 10.44) ---
def compute_J_M_eigenvalues(eigenvalues, M):
    """Compute J_M as the sum of eigenvalues from j=M+1 to D (Equation 10.44)."""
    return sum(eigenvalues[M:])

J_M_eigen = compute_J_M_eigenvalues(eigenvalues, M)
print(f"J_{M} via Eigenvalues (Σ λ_j for j={M+1} to {D}): {J_M_eigen:.3f}")
print(f"Error between J_{M} and Eigenvalue Sum: {abs(J_M - J_M_eigen):.3f}")

# --- Verify Displacement Lies in Orthogonal Complement ---
def verify_orthogonal_complement(displacements, B_orthogonal):
    """Verify that x_n - x̃_n lies in the span of b_{M+1}, ..., b_D."""
    print("\nVerifying Displacement Vectors Lie in Orthogonal Complement:")
    for n, disp in enumerate(displacements, 1):
        print(f"\nDisplacement x_{n} - x̃_{n}: {disp}")
        # Project displacement onto the orthogonal complement basis
        for j, b in enumerate(B_orthogonal, M+1):
            coord = sum(disp[i] * b[i] for i in range(D))
            print(f"Coordinate along b{j}: {coord:.3f}")
        # Check orthogonality to the principal subspace
        for j, b in enumerate(B, 1):
            dot = sum(disp[i] * b[i] for i in range(D))
            print(f"Dot product with b{j} (should be 0): {dot:.3e}")

# --- Main Demonstration ---
def demonstrate_pca_basis():
    """Demonstrate finding the optimal basis in PCA (Section 10.3.3)."""
    print("=== PCA Basis of Principal Subspace Demonstration ===")
    print("Section 10.3.3: Finding the Basis of the Principal Subspace\n")
    
    print(f"Optimal Basis for Principal Subspace (M={M}): {B}")
    print(f"Reconstruction Error J_{M}: {J_M:.3f}")
    print(f"J_{M} via Eigenvalues: {J_M_eigen:.3f}")
    
    # Verify displacement vectors
    B_orthogonal = basis_vectors[M:]
    verify_orthogonal_complement(displacements, B_orthogonal)

# --- Plotting Function ---
def plot_pca_basis(X, X_tilde, B, displacements):
    """Plot the centered data, reconstructed data, principal subspace, and displacement vectors."""
    plt.figure(figsize=(8, 6))
    
    # Plot centered data
    cent_x = [x[0] for x in X]
    cent_y = [x[1] for x in X]
    plt.scatter(cent_x, cent_y, color='blue', label='Centered Data (x_n)', alpha=0.7)
    
    # Plot reconstructed data
    recon_x = [x[0] for x in X_tilde]
    recon_y = [x[1] for x in X_tilde]
    plt.scatter(recon_x, recon_y, color='orange', label='Reconstructed Data (x̃_n)', alpha=0.7)
    
    # Plot displacement vectors (x_n - x̃_n)
    for n in range(N):
        plt.arrow(X_tilde[n][0], X_tilde[n][1], 
                  displacements[n][0], displacements[n][1],
                  color='red', alpha=0.5, width=0.02,
                  label='Displacement (x_n - x̃_n)' if n == 0 else "")
    
    # Plot the principal subspace (line spanned by b1)
    scale = 3
    b1 = B[0]
    plt.plot([-scale * b1[0], scale * b1[0]], [-scale * b1[1], scale * b1[1]],
             color='green', label='Principal Subspace (U)')
    
    # Plot the orthogonal complement (line spanned by b2)
    b2 = basis_vectors[1]
    plt.plot([-scale * b2[0], scale * b2[0]], [-scale * b2[1], scale * b2[1]],
             color='purple', linestyle='--', label='Orthogonal Complement (U⊥)')
    
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('PCA: Finding the Basis of the Principal Subspace')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PCA Basis of Principal Subspace Demonstration with Matplotlib Plotting")
    print("=" * 60)

    # Demonstrate PCA
    demonstrate_pca_basis()

    # Plot the results
    print("\nPlotting PCA Results")
    plot_pca_basis(X, X_tilde, B, displacements)

    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Optimal Basis for Principal Subspace: {B}")
    print(f"• Reconstruction Error J_{M}: {J_M:.3f}")
    print("• Plotted centered data, reconstructed data, displacement vectors, principal subspace, and orthogonal complement")
```

##  Eigenvector Computation and Low-Rank Approximations

In the previous sections, we determined that the basis of the principal subspace consists of the eigenvectors associated with the largest eigenvalues of the data covariance matrix $ S $, defined as:

$$
S = \frac{1}{N} \sum_{n=1}^N x_n x_n^\top = \frac{1}{N} X X^\top, \tag{10.45}
$$

where $ X = [x_1, \ldots, x_N] \in \mathbb{R}^{D \times N} $ is the data matrix:

$$
X \in \mathbb{R}^{D \times N}, \tag{10.46}
$$

with each column $ x_n \in \mathbb{R}^D $. Note that $ X $ is the transpose of the "typical" data matrix as defined in some texts (Bishop, 2006; Murphy, 2012), where rows usually represent data points.

### Computing Eigenvectors of $ S $

To obtain the eigenvalues and eigenvectors of $ S $, we can use one of two approaches:

1. **Eigendecomposition**: Perform an eigendecomposition of $ S $ (Section 4.2) to directly compute its eigenvalues and eigenvectors.
2. **Singular Value Decomposition (SVD)**: Use SVD on the data matrix $ X $ (Section 4.5) to indirectly compute the eigenvalues and eigenvectors of $ S $.

Since $ S $ is symmetric and can be factored as $ S = \frac{1}{N} X X^\top $, we can leverage SVD to find the eigenvectors. The SVD of $ X $ is:

$$
X = U \Sigma V^\top, \tag{10.47}
$$

where:
- $ U \in \mathbb{R}^{D \times D} $ and $ V^\top \in \mathbb{R}^{N \times N} $ are orthogonal matrices,
- $ \Sigma \in \mathbb{R}^{D \times N} $ is a rectangular diagonal matrix with non-negative singular values $ \sigma_{ii} \geq 0 $ on the diagonal and zeros elsewhere.

The dimensions of the matrices in the SVD are:
- $ X $: $ D \times N $,
- $ U $: $ D \times D $,
- $ \Sigma $: $ D \times N $,
- $ V^\top $: $ N \times N $.

Substituting the SVD into the expression for $ S $:

$$
S = \frac{1}{N} X X^\top = \frac{1}{N} (U \Sigma V^\top) (V \Sigma^\top U^\top) = \frac{1}{N} U \Sigma (V^\top V) \Sigma^\top U^\top = \frac{1}{N} U (\Sigma \Sigma^\top) U^\top, \tag{10.48}
$$

since $ V^\top V = I $ (because $ V $ is orthogonal). This shows that $ S $ has the form of an eigendecomposition:

$$
S = U \left( \frac{1}{N} \Sigma \Sigma^\top \right) U^\top,
$$

where:
- The columns of $ U $ are the eigenvectors of $ S $ (and thus of $ X X^\top $, up to the factor $ \frac{1}{N} $),
- The eigenvalues $ \lambda_d $ of $ S $ are the diagonal entries of $ \frac{1}{N} \Sigma \Sigma^\top $.

From Section 4.5, the eigenvalues $ \lambda_d $ of $ S $ are related to the singular values $ \sigma_d $ of $ X $:

$$
\lambda_d = \frac{\sigma_d^2}{N}, \tag{10.49}
$$

where $ \sigma_d $ are the singular values on the diagonal of $ \Sigma $. This relationship connects the maximum-variance perspective of PCA (Section 10.2) to the SVD, as the principal components (eigenvectors of $ S $) are the columns of $ U $, and the variances (eigenvalues) are determined by the singular values of $ X $.

## PCA Using Low-Rank Matrix Approximations

To maximize the variance of the projected data (or equivalently, minimize the average squared reconstruction error), PCA selects the eigenvectors corresponding to the $ M $ largest eigenvalues of $ S $. From Equation (10.48), these eigenvectors are the columns of $ U $. Thus, we identify $ U $ (or a subset of its columns) as the projection matrix $ B $ from Equation (10.3), which projects the data onto an $ M $-dimensional subspace.

The Eckart-Young theorem (Theorem 4.25 in Section 4.6) provides a direct method to compute this low-dimensional representation via a low-rank approximation. Consider the best rank-$ M $ approximation of the data matrix $ X $:

$$
\tilde{X}_M := \text{argmin}_{\text{rk}(A) \leq M} \lVert X - A \rVert_2 \in \mathbb{R}^{D \times N}, \tag{10.50}
$$

where $ \lVert \cdot \rVert_2 $ is the spectral norm (defined in Equation 4.93). According to the Eckart-Young theorem, the optimal rank-$ M $ approximation $ \tilde{X}_M $ is obtained by truncating the SVD of $ X $ to the top $ M $ singular values:

$$
\tilde{X}_M = U_M \Sigma_M V_M^\top \in \mathbb{R}^{D \times N}, \tag{10.51}
$$

where:
- $ U_M = [u_1, \ldots, u_M] \in \mathbb{R}^{D \times M} $ contains the first $ M $ columns of $ U $,
- $ \Sigma_M \in \mathbb{R}^{M \times M} $ is a diagonal matrix with the $ M $ largest singular values of $ X $,
- $ V_M = [v_1, \ldots, v_M] \in \mathbb{R}^{N \times M} $ contains the first $ M $ columns of $ V $,
- $ V_M^\top \in \mathbb{R}^{M \times N} $.

The matrices $ U_M $ and $ V_M $ are orthogonal (i.e., $ U_M^\top U_M = I $, $ V_M^\top V_M = I $), and $ \tilde{X}_M $ is the best rank-$ M $ approximation of $ X $ in the spectral norm. In the context of PCA, the columns of $ U_M $ are the principal components (basis vectors of the principal subspace), and the projection of the data onto this subspace is achieved by multiplying $ X $ by $ U_M U_M^\top $, effectively reconstructing the data with rank $ M $.

```python
import math
import matplotlib.pyplot as plt

# --- Define the Dataset ---
data = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N = len(data)
D = len(data[0])
M = 1  # Project onto a 1D subspace

# --- Center the Data ---
def compute_mean(data):
    """Compute the mean of the data across each dimension."""
    mean = [0.0] * D
    for x in data:
        for i in range(D):
            mean[i] += x[i] / N
    return mean

def center_data(data, mean):
    """Center the data by subtracting the mean."""
    centered_data = []
    for x in data:
        centered_x = [x[i] - mean[i] for i in range(D)]
        centered_data.append(centered_x)
    return centered_data

mean = compute_mean(data)
X_list = center_data(data, mean)
print("Original Data:", data)
print("Mean:", mean)
print("Centered Data (X):", X_list)

# Convert X_list to matrix form X (D x N) as per Equation 10.46
X = [[X_list[n][d] for n in range(N)] for d in range(D)]
print("\nData Matrix X (D x N):")
for row in X:
    print(row)

# --- Eigendecomposition of S (Equation 10.45) ---
def compute_covariance_matrix(X_list):
    """Compute S = (1/N) X X^T."""
    S = [[0.0 for _ in range(D)] for _ in range(D)]
    for x in X_list:
        for i in range(D):
            for j in range(D):
                S[i][j] += (x[i] * x[j]) / N
    return S

def compute_eigenvalues_2x2(S):
    """Compute eigenvalues of a 2x2 matrix."""
    a, b = S[0][0], S[0][1]
    c, d = S[1][0], S[1][1]
    trace = a + d
    det = a * d - b * c
    discriminant = trace**2 - 4 * det
    sqrt_disc = math.sqrt(discriminant)
    lambda1 = (trace + sqrt_disc) / 2
    lambda2 = (trace - sqrt_disc) / 2
    return lambda1, lambda2  # Returns in descending order

def compute_eigenvector_2x2(S, eigenvalue):
    """Compute the eigenvector for a given eigenvalue (normalized)."""
    a, b = S[0][0], S[0][1]
    c, d = S[1][0], S[1][1]
    if abs(a - eigenvalue) > 1e-10:
        v1 = -b / (a - eigenvalue)
        v2 = 1.0
    else:
        v1 = 1.0
        v2 = -(d - eigenvalue) / c if abs(c) > 1e-10 else 0.0
    norm = math.sqrt(v1**2 + v2**2)
    return [v1 / norm, v2 / norm]

S = compute_covariance_matrix(X_list)
print("\nCovariance Matrix S:")
for row in S:
    print(row)

eigenvalues = compute_eigenvalues_2x2(S)
eigenvectors = [compute_eigenvector_2x2(S, eigenvalues[0]),
                compute_eigenvector_2x2(S, eigenvalues[1])]
print(f"\nEigendecomposition of S:")
print(f"Eigenvalues: λ1 = {eigenvalues[0]:.3f}, λ2 = {eigenvalues[1]:.3f}")
print(f"Eigenvectors: u1 = {eigenvectors[0]}, u2 = {eigenvectors[1]}")

# --- SVD of X (Equation 10.47) ---
def matrix_multiply(A, B, A_rows, A_cols, B_cols):
    """Multiply two matrices."""
    result = [[0.0 for _ in range(B_cols)] for _ in range(A_rows)]
    for i in range(A_rows):
        for j in range(B_cols):
            for k in range(A_cols):
                result[i][j] += A[i][k] * B[k][j]
    return result

# Compute X^T X (N x N)
XT = [[X[d][n] for d in range(D)] for n in range(N)]
XTX = matrix_multiply(XT, X, N, D, N)
print("\nX^T X (N x N):")
for row in XTX:
    print([f"{val:.3f}" for val in row])

# Eigenvalues of X^T X are N times the eigenvalues of S
singular_values_squared = [N * ev for ev in eigenvalues]
singular_values = [math.sqrt(max(0, sv2)) for sv2 in singular_values_squared]

# U is the eigenvectors of X X^T (already computed as eigenvectors of S)
U = eigenvectors  # U = [u1, u2]

# Compute V by solving X v = σ u for each non-zero singular value
V = []
for d in range(min(D, N)):
    if singular_values[d] > 1e-10:
        u = U[d]
        sigma = singular_values[d]
        # X v = σ u  =>  v = (1/σ) X^T u
        v = [0.0] * N
        for n in range(N):
            for i in range(D):
                v[n] += X[i][n] * u[i] / sigma
        # Normalize v
        norm = math.sqrt(sum(vi**2 for vi in v))
        if norm > 1e-10:
            v = [vi / norm for vi in v]
            V.append(v)

# Ensure V has N vectors by padding with orthogonal vectors
# Use Gram-Schmidt to orthogonalize remaining vectors
def gram_schmidt(V, N):
    """Orthogonalize vectors using Gram-Schmidt and pad to N vectors."""
    # Start with the vectors we have
    basis = V.copy()
    num_vectors = len(basis)
    
    # Pad with standard basis vectors and orthogonalize
    for i in range(num_vectors, N):
        # Start with a standard basis vector
        v = [0.0] * N
        v[i] = 1.0
        # Orthogonalize against previous vectors
        for u in basis:
            dot = sum(v[j] * u[j] for j in range(N))
            for j in range(N):
                v[j] -= dot * u[j]
        # Normalize
        norm = math.sqrt(sum(vj**2 for vj in v))
        if norm > 1e-10:
            v = [vj / norm for vj in v]
            basis.append(v)
    return basis

V = gram_schmidt(V, N)
print("\nSVD of X:")
print(f"Singular Values: σ1 = {singular_values[0]:.3f}, σ2 = {singular_values[1]:.3f}")
print("U (from X X^T):")
for u in U:
    print(u)
print("V:")
for v in V:
    print([f"{vi:.3f}" for vi in v])

# Verify λ_d = σ_d^2 / N (Equation 10.49)
print("\nVerifying Eigenvalues vs Singular Values (Equation 10.49):")
for d in range(min(D, N)):
    lambda_from_sv = (singular_values[d]**2) / N
    print(f"λ_{d+1} from S: {eigenvalues[d]:.3f}, σ_{d+1}^2/N: {lambda_from_sv:.3f}")

# --- Low-Rank Approximation (Equation 10.51) ---
U_M = U[:M]
Sigma_M = [[singular_values[m] if i == m else 0 for i in range(M)] for m in range(M)]
V_M = [V[m][:N] for m in range(M)]
V_M_T = [[V_M[m][n] for n in range(N)] for m in range(M)]

# Compute X̃_M = U_M Σ_M V_M^T
U_Sigma = matrix_multiply(U_M, Sigma_M, D, M, M)
X_tilde_M = matrix_multiply(U_Sigma, V_M_T, D, M, N)
print("\nRank-M Approximation X̃_M (D x N):")
for row in X_tilde_M:
    print([f"{val:.3f}" for val in row])

# Convert X̃_M back to list form for plotting
X_tilde_M_list = [[X_tilde_M[d][n] for d in range(D)] for n in range(N)]

# --- Compute Reconstruction Error ---
def compute_reconstruction_error(X, X_tilde):
    """Compute the Frobenius norm ||X - X̃||_F as a proxy for spectral norm."""
    diff = [[X[d][n] - X_tilde[d][n] for n in range(N)] for d in range(D)]
    frobenius_norm = math.sqrt(sum(sum(diff[d][n]**2 for n in range(N)) for d in range(D)))
    return frobenius_norm

error = compute_reconstruction_error(X, X_tilde_M)
print(f"\nApproximation Error (Frobenius norm of X - X̃_M): {error:.3f}")

# --- Main Demonstration ---
def demonstrate_pca_svd_lowrank():
    """Demonstrate PCA using SVD and low-rank approximation (Section 10.4)."""
    print("=== PCA with SVD and Low-Rank Approximation Demonstration ===")
    print("Section 10.4: Eigenvector Computation and Low-Rank Approximations\n")
    
    print(f"Principal Components (from U_M, M={M}):")
    for m, u in enumerate(U_M, 1):
        print(f"u_{m}: {u}")
    print(f"Reconstruction Error (Frobenius norm): {error:.3f}")

# --- Plotting Function ---
def plot_pca_lowrank(X_list, X_tilde_M_list):
    """Plot the original centered data and low-rank approximated data."""
    plt.figure(figsize=(8, 6))
    
    cent_x = [x[0] for x in X_list]
    cent_y = [x[1] for x in X_list]
    plt.scatter(cent_x, cent_y, color='blue', label='Centered Data (x_n)', alpha=0.7)
    
    approx_x = [x[0] for x in X_tilde_M_list]
    approx_y = [x[1] for x in X_tilde_M_list]
    plt.scatter(approx_x, approx_y, color='orange', label='Approximated Data (x̃_n)', alpha=0.7)
    
    for n in range(N):
        plt.plot([X_list[n][0], X_tilde_M_list[n][0]], [X_list[n][1], X_tilde_M_list[n][1]],
                 'r--', alpha=0.5, label='Difference' if n == 0 else "")
    
    scale = 3
    u1 = U_M[0]
    plt.plot([-scale * u1[0], scale * u1[0]], [-scale * u1[1], scale * u1[1]],
             color='green', label='Principal Subspace')
    
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('PCA: Low-Rank Approximation via SVD (M=1)')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PCA with SVD and Low-Rank Approximation Demonstration")
    print("=" * 60)

    demonstrate_pca_svd_lowrank()
    print("\nPlotting PCA Results")
    plot_pca_lowrank(X_list, X_tilde_M_list)
    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Principal Components (U_M): {U_M}")
    print(f"• Approximation Error (Frobenius norm): {error:.3f}")
    print("• Plotted centered data, approximated data, and principal subspace")
```

## Practical Aspects

Finding eigenvalues and eigenvectors is a fundamental task in many machine learning methods that rely on matrix decompositions, such as PCA. In theory, as discussed in Section 4.2, we can find eigenvalues by solving for the roots of the characteristic polynomial of a matrix. However, for matrices larger than $ 4 \times 4 $, this approach becomes impractical. The characteristic polynomial would be of degree 5 or higher, and the Abel-Ruffini theorem (Ruffini, 1799; Abel, 1826) states that there is no general algebraic solution for polynomials of degree 5 or greater. Therefore, in practice, we rely on iterative methods to compute eigenvalues or singular values. These iterative methods are implemented in modern linear algebra packages and are designed to be efficient.

In many applications, including PCA as presented in this chapter, we are often interested in only a few eigenvectors—those corresponding to the largest eigenvalues. Computing a full eigendecomposition (or SVD) and then discarding most of the eigenvectors (those associated with smaller eigenvalues) would be computationally wasteful. Fortunately, iterative methods that directly target the first few eigenvectors are more efficient than a full decomposition. In the extreme case, where we need only the first eigenvector (the one associated with the largest eigenvalue), a simple and efficient method called *power iteration* can be used.

### Power Iteration

Power iteration is an iterative algorithm that finds the dominant eigenvector of a matrix. Starting with a random vector $ x_0 $ that is not in the null space of the covariance matrix $ S $, the power iteration proceeds as follows:

$$
x_{k+1} = \frac{S x_k}{\lVert S x_k \rVert}, \quad k = 0, 1, \ldots, \tag{10.52}
$$

In each iteration:
- The vector $ x_k $ is multiplied by $ S $,
- The result $ S x_k $ is normalized to ensure $ \lVert x_{k+1} \rVert = 1 $.

If $ S $ is invertible, it is sufficient to ensure that $ x_0 \neq 0 $ to avoid the null space. Under these conditions, the sequence of vectors $ \{ x_k \} $ converges to the eigenvector associated with the largest eigenvalue of $ S $. The power iteration method is notably used in the original Google PageRank algorithm (Page et al., 1999) to rank web pages based on their hyperlink structure.

## PCA in High Dimensions

To perform PCA, we need to compute the data covariance matrix and its eigendecomposition. In $ D $ dimensions, the covariance matrix $ S $ is a $ D \times D $ matrix. Computing the eigenvalues and eigenvectors of $ S $ is computationally expensive, with a time complexity that scales cubically with $ D $ (i.e., $ O(D^3) $). As a result, the PCA approach described earlier becomes infeasible in very high dimensions. For example, if our data points $ x_n $ are images with 10,000 pixels (e.g., $ 100 \times 100 $ pixel images), $ D = 10,000 $, and we would need to compute the eigendecomposition of a $ 10,000 \times 10,000 $ covariance matrix, which is computationally prohibitive.

In the following, we address this challenge for the case where the number of data points is much smaller than the dimensionality, i.e., $ N \ll D $. This scenario is common in many high-dimensional datasets, such as image or genomic data, where the number of features (dimensions) far exceeds the number of samples.

### Reformulating PCA for High Dimensions

Assume we have a centered dataset $ x_1, \ldots, x_N $, where each $ x_n \in \mathbb{R}^D $. The data covariance matrix is:

$$
S = \frac{1}{N} X X^\top \in \mathbb{R}^{D \times D}, \tag{10.53}
$$

where $ X = [x_1, \ldots, x_N] $ is a $ D \times N $ matrix, with each column representing a data point. We assume $ N \ll D $, meaning the number of data points is much smaller than the dimensionality of the data.

Since $ X $ is $ D \times N $, the rank of $ X X^\top $ (and thus $ S $) is at most $ \min(D, N) = N $ (assuming no duplicate data points). Therefore, $ S $ has at most $ N $ non-zero eigenvalues, and there are $ D - N + 1 $ eigenvalues that are zero. Intuitively, this indicates redundancy in the covariance structure, as the effective dimensionality of the data is limited by the number of samples $ N $.

To exploit this redundancy, we can transform the problem of computing the eigendecomposition of the $ D \times D $ matrix $ S $ into a more manageable problem involving an $ N \times N $ matrix, whose eigenvalues are all positive (since we focus on the non-zero eigenvalues). This approach significantly reduces the computational burden when $ N \ll D $.

### Deriving the Reduced Eigendecomposition

In PCA, we solve the eigenvector equation for the principal subspace basis vectors $ b_m $:

$$
S b_m = \lambda_m b_m, \quad m = 1, \ldots, M, \tag{10.54}
$$

where $ b_m $ is a basis vector of the principal subspace, and $ \lambda_m $ is the corresponding eigenvalue. Substituting $ S = \frac{1}{N} X X^\top $ from Equation (10.53) into this equation:

$$
\frac{1}{N} X X^\top b_m = \lambda_m b_m. \tag{10.55}
$$

To reformulate this, multiply both sides of Equation (10.55) by $ X^\top \in \mathbb{R}^{N \times D} $ from the left:

$$
\frac{1}{N} X^\top (X X^\top b_m) = \lambda_m X^\top b_m.
$$

Define $ c_m := X^\top b_m \in \mathbb{R}^N $. Then the equation becomes:

$$
\frac{1}{N} (X^\top X) (X^\top b_m) = \lambda_m (X^\top b_m) \implies \frac{1}{N} (X^\top X) c_m = \lambda_m c_m, \tag{10.56}
$$

where $ X^\top X \in \mathbb{R}^{N \times N} $ is an $ N \times N $ matrix. This is now an eigenvalue problem for the smaller matrix $ X^\top X $, which is much more tractable when $ N \ll D $. The eigenvectors $ c_m $ of $ X^\top X $ can be used to recover the eigenvectors $ b_m $ of $ S $, as we will explore further.

### Reduced Eigendecomposition (Continued)

From Equation (10.56), we obtained a new eigenvalue equation for the $ N \times N $ matrix $ \frac{1}{N} X^\top X $:

$$
\frac{1}{N} (X^\top X) c_m = \lambda_m c_m, \tag{10.56}
$$

where $ c_m = X^\top b_m \in \mathbb{R}^N $, and $ \lambda_m $ remains the eigenvalue. This confirms the result from Section 4.5.3 that the non-zero eigenvalues of $ X X^\top $ (and thus $ S = \frac{1}{N} X X^\top $) are equal to the non-zero eigenvalues of $ X^\top X $. Since we assume no duplicate data points, the matrix $ X^\top X \in \mathbb{R}^{N \times N} $ has rank $ N $ and is invertible. Therefore, $ \frac{1}{N} X^\top X $ has the same non-zero eigenvalues as the data covariance matrix $ S $, but it is only an $ N \times N $ matrix. This allows us to compute its eigenvalues and eigenvectors much more efficiently than for the original $ D \times D $ covariance matrix $ S $, especially when $ N \ll D $.

#### Recovering the Eigenvectors of $ S $

Now that we have the eigenvectors $ c_m $ of $ \frac{1}{N} X^\top X $, we need to recover the eigenvectors $ b_m $ of $ S $ to perform PCA. Left-multiply Equation (10.56) by $ X \in \mathbb{R}^{D \times N} $:

$$
X \left( \frac{1}{N} (X^\top X) c_m \right) = \lambda_m (X c_m) \implies \frac{1}{N} X (X^\top X) c_m = \lambda_m (X c_m).
$$

Since $ S = \frac{1}{N} X X^\top $, and $ X c_m = X (X^\top b_m) = (X X^\top) b_m $, we substitute:

$$
\frac{1}{N} (X X^\top) (X c_m) = \lambda_m (X c_m) \implies S (X c_m) = \lambda_m (X c_m), \tag{10.57}
$$

This shows that $ X c_m $ is an eigenvector of $ S $ corresponding to the eigenvalue $ \lambda_m $. Thus, the eigenvectors $ b_m $ of $ S $ can be recovered as $ b_m = X c_m $, up to normalization.

**Remark**: To apply the PCA algorithm as discussed in Section 10.6, we must normalize the eigenvectors $ X c_m $ of $ S $ to have norm 1, i.e., $ b_m = \frac{X c_m}{\lVert X c_m \rVert} $, ensuring they form an orthonormal basis. $ \diamond $

## 10.6 Key Steps of PCA in Practice

We now outline the practical steps of PCA using a running example with a two-dimensional dataset (\( D = 2 \)), which we project onto a one-dimensional subspace (\( M = 1 \)). The process is illustrated in Figure 10.11.

### Step 1: Mean Subtraction

Center the data by computing the mean $ \mu $ of the dataset and subtracting it from each data point:

$$
x_n \leftarrow x_n - \mu, \quad \text{for all } n = 1, \ldots, N,
$$

where $ \mu = \frac{1}{N} \sum_{n=1}^N x_n $. This ensures the dataset has mean zero, as shown in Figure 10.11(b). While mean subtraction is not strictly necessary for PCA, it reduces the risk of numerical issues.

### Step 2: Standardization

Standardize the data by dividing each dimension of the data points by the standard deviation $ \sigma_d $ of the dataset in that dimension:

$$
x_n^{(d)} \leftarrow \frac{x_n^{(d)}}{\sigma_d}, \quad d = 1, \ldots, D,
$$

where $ x_n^{(d)} $ is the $ d $-th component of $ x_n $, and $ \sigma_d = \sqrt{\frac{1}{N} \sum_{n=1}^N (x_n^{(d)} - \mu_d)^2} $ (assuming the data is already centered, so $ \mu_d = 0 $). After this step, the data is unit-free and has variance 1 along each axis, as indicated by the arrows in Figure 10.11(c).

### Step 3: Eigendecomposition of the Covariance Matrix

Compute the data covariance matrix $ S = \frac{1}{N} X X^\top $, where $ X $ is the centered and standardized data matrix. Then, perform an eigendecomposition to find the eigenvalues and eigenvectors of $ S $. Since $ S $ is symmetric, the spectral theorem (Theorem 4.15) guarantees that we can find an orthonormal basis (ONB) of eigenvectors. In Figure 10.11(d), the eigenvectors are scaled by the magnitude of their corresponding eigenvalues. The eigenvector associated with the largest eigenvalue (the longer vector) spans the principal subspace $ U $, and the covariance matrix is represented by an ellipse.

### Step 4: Projection

Project a data point $ x^* \in \mathbb{R}^D $ onto the principal subspace $ U $. First, standardize $ x^* $ using the mean $ \mu_d $ and standard deviation $ \sigma_d $ of the training data in each dimension:

$$
x^{*(d)} \leftarrow \frac{x^{*(d)} - \mu_d}{\sigma_d}, \quad d = 1, \ldots, D, \tag{10.58}
$$

where $ x^{*(d)} $ is the $ d $-th component of $ x^* $. Then, compute the projection onto the principal subspace:

$$
\tilde{x}^* = B B^\top x^*, \tag{10.59}
$$

where $ B $ is the matrix whose columns are the $ M $ eigenvectors of $ S $ corresponding to the largest eigenvalues (i.e., the basis vectors of the principal subspace). The coordinates of the projection with respect to this basis are:

$$
z^* = B^\top x^*. \tag{10.60}
$$

PCA typically returns the coordinates $ z^* $ (Equation 10.60), not the projection $ \tilde{x}^* $. Since the dataset was standardized, $ \tilde{x}^* $ is in the standardized space.

### Step 5: Mapping Back to the Original Data Space

To obtain the projection in the original data space (before standardization), undo the standardization by reversing Equation (10.58):

$$
\tilde{x}^{*(d)} \leftarrow \tilde{x}^{*(d)} \sigma_d + \mu_d, \quad d = 1, \ldots, D. \tag{10.61}
$$

This step multiplies each component of $ \tilde{x}^* $ by the standard deviation $ \sigma_d $ and adds the mean $ \mu_d $, mapping the projection back to the original space, as shown in Figure 10.11(f).

We apply PCA to the MNIST digits dataset, which contains 60,000 examples of handwritten digits (0 through 9). Each digit is a $ 28 \times 28 $ image, so each image is a vector $ x \in \mathbb{R}^{784} $ (since $ 28 \times 28 = 784 $ pixels). Examples of these digits are shown in Figure 10.3.

For illustration, we focus on a subset of the MNIST dataset containing only the digit "8", with 5,389 training images. We compute the principal subspace as described in this chapter and use the learned projection matrix to reconstruct a set of test images, as shown in Figure 10.12.

### Reconstruction Error Analysis

Figure 10.13 plots the average squared reconstruction error as a function of the number of principal components $ M $:

$$
\frac{1}{N} \sum_{n=1}^N \lVert x_n - \tilde{x}_n \rVert^2 = \sum_{i=M+1}^D \lambda_i, \tag{10.62}
$$

where $ \lambda_i $ are the eigenvalues of the data covariance matrix. The plot shows that the reconstruction error decreases rapidly as $ M $ increases, with only marginal improvements beyond a certain point. This aligns with observations in Figure 10.5, where most of the variance in the projected data is captured by the first few principal components.

For the digit "8" dataset:
- With about 550 PCs, the reconstruction error is nearly zero, indicating near-perfect reconstruction.
- Some pixels (e.g., around the boundaries) show no variation across the dataset (always black), so fewer than 784 PCs may suffice for practical purposes.

### Interpretation

The rapid drop-off in reconstruction error demonstrates the effectiveness of PCA for dimensionality reduction. By using only a small number of principal components, we can capture most of the variance in the data, achieving significant compression while retaining the essential structure of the digits.

```python
import math
import matplotlib.pyplot as plt

# --- Define the Dataset ---
data = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N = len(data)
D = len(data[0])
M = 1  # Project onto a 1D subspace

# New data point to project (as per Section 10.6, Step 4)
x_star = [2.5, 2.5]

print("Original Data:", data)
print("New Data Point x*:", x_star)

# --- Step 1: Mean Subtraction ---
def compute_mean(data):
    """Compute the mean of the data across each dimension."""
    mean = [0.0] * D
    for x in data:
        for i in range(D):
            mean[i] += x[i] / N
    return mean

def center_data(data, mean):
    """Center the data by subtracting the mean."""
    centered_data = []
    for x in data:
        centered_x = [x[i] - mean[i] for i in range(D)]
        centered_data.append(centered_x)
    return centered_data

mean = compute_mean(data)
X_centered = center_data(data, mean)
print("\nStep 1: Mean Subtraction")
print("Mean (μ):", mean)
print("Centered Data (X_centered):", X_centered)

# Center the new data point x*
x_star_centered = [x_star[i] - mean[i] for i in range(D)]
print("Centered x*:", x_star_centered)

# --- Step 2: Standardization ---
def compute_std_dev(data):
    """Compute the standard deviation for each dimension (data is centered)."""
    std_dev = [0.0] * D
    for x in data:
        for i in range(D):
            std_dev[i] += x[i]**2 / N
    return [math.sqrt(max(0, s)) for s in std_dev]

def standardize_data(data, std_dev):
    """Standardize the data by dividing by the standard deviation."""
    standardized_data = []
    for x in data:
        standardized_x = [x[i] / std_dev[i] if std_dev[i] > 1e-10 else 0.0 for i in range(D)]
        standardized_data.append(standardized_x)
    return standardized_data

std_dev = compute_std_dev(X_centered)
X_standardized = standardize_data(X_centered, std_dev)
print("\nStep 2: Standardization")
print("Standard Deviations (σ):", std_dev)
print("Standardized Data (X_standardized):", X_standardized)

# Standardize the new data point x*
x_star_standardized = [x_star_centered[i] / std_dev[i] if std_dev[i] > 1e-10 else 0.0 for i in range(D)]
print("Standardized x*:", x_star_standardized)

# Convert X_standardized to matrix form X (D x N)
X = [[X_standardized[n][d] for n in range(N)] for d in range(D)]

# --- Section 10.5: High-Dimensional PCA Approach ---
def matrix_multiply(A, B, A_rows, A_cols, B_cols):
    """Multiply two matrices."""
    result = [[0.0 for _ in range(B_cols)] for _ in range(A_rows)]
    for i in range(A_rows):
        for j in range(B_cols):
            for k in range(A_cols):
                result[i][j] += A[i][k] * B[k][j]
    return result

# Compute X^T X (N x N matrix)
XT = [[X[d][n] for d in range(D)] for n in range(N)]
XTX = matrix_multiply(XT, X, N, D, N)
# Scale by 1/N to get (1/N) X^T X
for i in range(N):
    for j in range(N):
        XTX[i][j] /= N
print("\n(1/N) X^T X (N x N):")
for row in XTX:
    print([f"{val:.3f}" for val in row])

# Compute S = (1/N) X X^T (D x D) to get eigenvalues
S = matrix_multiply(X, XT, D, N, D)  # Fixed dimensions: X is D x N, XT is N x D, result is D x D
for i in range(D):
    for j in range(D):
        S[i][j] /= N
print("\nS = (1/N) X X^T (D x D):")
for row in S:
    print([f"{val:.3f}" for val in row])

def compute_eigenvalues_2x2(matrix):
    """Compute eigenvalues of a 2x2 matrix."""
    a, b = matrix[0][0], matrix[0][1]
    c, d = matrix[1][0], matrix[1][1]
    trace = a + d
    det = a * d - b * c
    discriminant = trace**2 - 4 * det
    sqrt_disc = math.sqrt(discriminant)
    lambda1 = (trace + sqrt_disc) / 2
    lambda2 = (trace - sqrt_disc) / 2
    return lambda1, lambda2

def compute_eigenvector_2x2(matrix, eigenvalue):
    """Compute the eigenvector for a given eigenvalue (normalized)."""
    a, b = matrix[0][0], matrix[0][1]
    c, d = matrix[1][0], matrix[1][1]
    if abs(a - eigenvalue) > 1e-10:
        v1 = -b / (a - eigenvalue)
        v2 = 1.0
    else:
        v1 = 1.0
        v2 = -(d - eigenvalue) / c if abs(c) > 1e-10 else 0.0
    norm = math.sqrt(v1**2 + v2**2)
    return [v1 / norm, v2 / norm]

eigenvalues_S = compute_eigenvalues_2x2(S)
c1_S = compute_eigenvector_2x2(S, eigenvalues_S[0])
c2_S = compute_eigenvector_2x2(S, eigenvalues_S[1])
print("\nEigenvalues of S (and thus of (1/N) X^T X):")
print(f"λ1 = {eigenvalues_S[0]:.3f}, λ2 = {eigenvalues_S[1]:.3f}")

# Eigenvectors of S (D x D)
U = [c1_S, c2_S]
# Compute c_m = X^T b_m and b_m = X c_m to recover eigenvectors of S
b_m = []
for m in range(M):
    cm = [0.0] * N
    # c_m = X^T U[m]
    for n in range(N):
        for d in range(D):
            cm[n] += X[d][n] * U[m][d]
    # b_m = X c_m
    bm = [0.0] * D
    for d in range(D):
        for n in range(N):
            bm[d] += X[d][n] * cm[n]
    # Normalize b_m
    norm = math.sqrt(sum(bi**2 for bi in bm))
    bm = [bi / norm if norm > 1e-10 else 0.0 for bi in bm]
    b_m.append(bm)

B = b_m
print("\nRecovered Eigenvectors of S (B, normalized):")
for m, bm in enumerate(B, 1):
    print(f"b_{m}: {bm}")

# --- Step 4: Projection ---
BBT = [[0.0 for _ in range(D)] for _ in range(D)]
for b in B:
    for i in range(D):
        for j in range(D):
            BBT[i][j] += b[i] * b[j]

def matrix_vector_multiply(matrix, vector):
    """Multiply a DxD matrix by a D-dimensional vector."""
    result = [0.0] * D
    for i in range(D):
        for j in range(D):
            result[i] += matrix[i][j] * vector[j]
    return result

x_tilde_star = matrix_vector_multiply(BBT, x_star_standardized)
print("\nStep 4: Projection")
print("Projection in standardized space (x̃*):", x_tilde_star)

z_star = [0.0] * M
for m in range(M):
    for d in range(D):
        z_star[m] += B[m][d] * x_star_standardized[d]
print("Coordinates z* in principal subspace:", z_star)

# --- Step 5: Map Back to Original Space ---
x_tilde_star_original = [0.0] * D
for d in range(D):
    x_tilde_star_original[d] = x_tilde_star[d] * std_dev[d] + mean[d]
print("\nStep 5: Mapping Back to Original Space")
print("Projection in original space:", x_tilde_star_original)

# --- Main Demonstration ---
def demonstrate_pca_highdim_practical():
    """Demonstrate PCA for high dimensions and practical steps (Sections 10.5-10.6)."""
    print("\n=== PCA High-Dimensional and Practical Steps Demonstration ===")
    print("Sections 10.5-10.6: PCA in High Dimensions and Practical Steps\n")
    
    print(f"Principal Subspace Basis (B, M={M}):")
    for m, b in enumerate(B, 1):
        print(f"b_{m}: {b}")
    print(f"Projected New Point in Original Space:", x_tilde_star_original)

# --- Plotting Function ---
def plot_pca_steps(data, X_centered, X_standardized, B, x_star, x_tilde_star_original):
    """Plot the steps of PCA as in Figure 10.11."""
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()

    # (a) Original dataset
    ax = axes[0]
    data_x = [x[0] for x in data]
    data_y = [x[1] for x in data]
    ax.scatter(data_x, data_y, color='blue', label='Data')
    ax.scatter([x_star[0]], [x_star[1]], color='red', label='x*', marker='x')
    ax.set_title('(a) Original Dataset')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    # (b) Centered data
    ax = axes[1]
    cent_x = [x[0] for x in X_centered]
    cent_y = [x[1] for x in X_centered]
    ax.scatter(cent_x, cent_y, color='blue', label='Centered Data')
    ax.scatter([x_star_centered[0]], [x_star_centered[1]], color='red', label='x*', marker='x')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    ax.set_title('(b) After Centering')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    # (c) Standardized data
    ax = axes[2]
    std_x = [x[0] for x in X_standardized]
    std_y = [x[1] for x in X_standardized]
    ax.scatter(std_x, std_y, color='blue', label='Standardized Data')
    ax.scatter([x_star_standardized[0]], [x_star_standardized[1]], color='red', label='x*', marker='x')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    ax.plot([-2, 2], [0, 0], 'g--', label='Variance 1 axes')
    ax.plot([0, 0], [-2, 2], 'g--')
    ax.set_title('(c) After Standardization')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    # (d) Eigendecomposition
    ax = axes[3]
    ax.scatter(std_x, std_y, color='blue', label='Standardized Data')
    ax.scatter([x_star_standardized[0]], [x_star_standardized[1]], color='red', label='x*', marker='x')
    scale = math.sqrt(eigenvalues_S[0]) * 2
    b1 = B[0]
    ax.plot([-scale * b1[0], scale * b1[0]], [-scale * b1[1], scale * b1[1]], 'g-', label='Principal Subspace')
    scale2 = math.sqrt(eigenvalues_S[1]) * 2
    b2 = U[1]
    ax.plot([-scale2 * b2[0], scale2 * b2[0]], [-scale2 * b2[1], scale2 * b2[1]], 'g--', label='Orthogonal')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    ax.set_title('(d) Eigendecomposition')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    # (e) Projection in standardized space
    ax = axes[4]
    ax.scatter(std_x, std_y, color='blue', label='Standardized Data')
    ax.scatter([x_star_standardized[0]], [x_star_standardized[1]], color='red', label='x*', marker='x')
    ax.scatter([x_tilde_star[0]], [x_tilde_star[1]], color='orange', label='x̃*', marker='o')
    ax.plot([x_star_standardized[0], x_tilde_star[0]], [x_star_standardized[1], x_tilde_star[1]], 'r--')
    ax.plot([-scale * b1[0], scale * b1[0]], [-scale * b1[1], scale * b1[1]], 'g-', label='Principal Subspace')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    ax.set_title('(e) Projection (Standardized)')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    # (f) Projection in original space
    ax = axes[5]
    ax.scatter(data_x, data_y, color='blue', label='Original Data')
    ax.scatter([x_star[0]], [x_star[1]], color='red', label='x*', marker='x')
    ax.scatter([x_tilde_star_original[0]], [x_tilde_star_original[1]], color='orange', label='x̃*', marker='o')
    ax.plot([x_star[0], x_tilde_star_original[0]], [x_star[1], x_tilde_star_original[1]], 'r--')
    b1_orig = [b1[d] * std_dev[d] for d in range(D)]
    scale_orig = 2
    ax.plot([mean[0] - scale_orig * b1_orig[0], mean[0] + scale_orig * b1_orig[0]],
            [mean[1] - scale_orig * b1_orig[1], mean[1] + scale_orig * b1_orig[1]], 'g-', label='Principal Subspace')
    ax.set_title('(f) Projection (Original Space)')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')

    plt.tight_layout()
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PCA High-Dimensional and Practical Steps Demonstration")
    print("=" * 60)

    demonstrate_pca_highdim_practical()
    print("\nPlotting PCA Steps (Figure 10.11)")
    plot_pca_steps(data, X_centered, X_standardized, B, x_star, x_tilde_star_original)
    print("\n" + "=" * 60)
    print("Summary of Results and Visualizations:")
    print(f"• Principal Subspace Basis (B): {B}")
    print(f"• Projected New Point (x̃* in original space): {x_tilde_star_original}")
    print("• Plotted PCA steps: original, centered, standardized, eigendecomposition, projection, and mapping back")
```
