## PCA: Multiple Perspectives - Mathematical Derivations

### Introduction

Principal Component Analysis (PCA) can be understood from multiple perspectives that all lead to the same mathematical solution. This notebook explores three key interpretations:

1. **Variance Maximization**: Finding directions of maximum variance in the data
2. **Reconstruction Error Minimization**: Minimizing the error when reconstructing original data
3. **Linear Auto-encoder**: Encoding and decoding data through linear transformations
4. **Information Theory**: Maximizing mutual information between original and compressed data



Fig.16 PCA can be viewed as a linear auto-encoder. It encodes the high-dimensional data x into a lower-dimensional representation (code) z ∈ RM and decodes z using a decoder. The decoded vector x̃ is the orthogonal projection of the original data x onto the M -dimensional principal subspace.

---

## 1. Variance Maximization Perspective

The first perspective seeks to find linear combinations of the original variables that have maximum variance.

**Problem Setup:**
- Data: $\mathbf{x} \in \mathbb{R}^D$ (high-dimensional)
- Projection: $\mathbf{z} = \mathbf{B}^T\mathbf{x} \in \mathbb{R}^M$ (low-dimensional)
- Goal: Maximize variance of projected data

**Mathematical Formulation:**

$$\max_{\mathbf{B}} \text{Var}(\mathbf{z}) = \max_{\mathbf{B}} \text{Var}(\mathbf{B}^T\mathbf{x})$$

Subject to orthonormality constraints: $\mathbf{B}^T\mathbf{B} = \mathbf{I}_M$

**Solution:**
The columns of $\mathbf{B}$ are the eigenvectors of the data covariance matrix $\mathbf{S}$ corresponding to the $M$ largest eigenvalues.

---

## 2. Reconstruction Error Minimization

The second perspective focuses on finding the best low-dimensional representation that minimizes reconstruction error.

**Problem Setup:**
- Original data: $\mathbf{x}_n \in \mathbb{R}^D$ for $n = 1, \ldots, N$
- Low-dimensional code: $\mathbf{z}_n = \mathbf{B}^T\mathbf{x}_n \in \mathbb{R}^M$
- Reconstruction: $\tilde{\mathbf{x}}_n = \mathbf{B}\mathbf{z}_n = \mathbf{B}\mathbf{B}^T\mathbf{x}_n$

**Objective Function:**

$$\min_{\mathbf{B}} \frac{1}{N}\sum_{n=1}^N \|\mathbf{x}_n - \tilde{\mathbf{x}}_n\|_2^2 = \min_{\mathbf{B}} \frac{1}{N}\sum_{n=1}^N \|\mathbf{x}_n - \mathbf{B}\mathbf{B}^T\mathbf{x}_n\|_2^2$$

This can be rewritten as:

$$\min_{\mathbf{B}} \frac{1}{N}\sum_{n=1}^N \|(\mathbf{I} - \mathbf{B}\mathbf{B}^T)\mathbf{x}_n\|_2^2$$

**Key Insight:** Minimizing reconstruction error is equivalent to maximizing the variance in the projected space.

---

## 3. Linear Auto-encoder Perspective

PCA can be viewed as a special case of a linear auto-encoder with tied weights.

### Auto-encoder Components:

**Encoder:** $\mathbf{z} = f_{\text{enc}}(\mathbf{x}) = \mathbf{B}^T\mathbf{x}$

**Decoder:** $\tilde{\mathbf{x}} = f_{\text{dec}}(\mathbf{z}) = \mathbf{B}\mathbf{z}$

**Complete Transformation:** $\tilde{\mathbf{x}} = \mathbf{B}\mathbf{B}^T\mathbf{x}$

### Loss Function:

$$\mathcal{L} = \frac{1}{N}\sum_{n=1}^N \|\mathbf{x}_n - \tilde{\mathbf{x}}_n\|_2^2 = \frac{1}{N}\sum_{n=1}^N \|\mathbf{x}_n - \mathbf{B}\mathbf{B}^T\mathbf{x}_n\|_2^2$$

### Properties:
- **Linear mappings**: Both encoder and decoder are linear transformations
- **Tied weights**: The decoder matrix $\mathbf{B}$ is the transpose of the encoder matrix $\mathbf{B}^T$
- **Orthogonal projection**: $\mathbf{B}\mathbf{B}^T$ is a projection matrix onto the $M$-dimensional principal subspace

### Extension to Nonlinear Auto-encoders:
When linear mappings are replaced with nonlinear functions (e.g., neural networks):
- **Deep Auto-encoder**: Uses deep neural networks for encoding and decoding
- **Recognition Network**: Alternative name for the encoder
- **Generator Network**: Alternative name for the decoder

---

## 4. Information Theory Perspective

From an information-theoretic viewpoint, PCA can be understood as maximizing the mutual information between the original data and its compressed representation.

### Key Concepts:

**Compression Interpretation:**
- The code $\mathbf{z}$ is a compressed version of the original data $\mathbf{x}$
- Reconstruction $\tilde{\mathbf{x}}$ is a "lossy" version of the original data
- Goal: Maximize correlation between original and compressed data

**Mutual Information Maximization:**

$$\max_{\mathbf{B}} I(\mathbf{x}; \mathbf{z}) = \max_{\mathbf{B}} I(\mathbf{x}; \mathbf{B}^T\mathbf{x})$$

where $I(\mathbf{x}; \mathbf{z})$ is the mutual information between $\mathbf{x}$ and $\mathbf{z}$.

**Connection to PCA:**
Maximizing mutual information under linear constraints leads to the same solution as the variance maximization approach.

---

## 5. Probabilistic PCA (PPCA)

### Model Specification:

**Generative Model:**
$$\mathbf{x} = \mathbf{B}\mathbf{z} + \boldsymbol{\mu} + \boldsymbol{\epsilon}$$

where:
- $\mathbf{z} \sim \mathcal{N}(\mathbf{0}, \mathbf{I}_M)$ (latent variables)
- $\boldsymbol{\epsilon} \sim \mathcal{N}(\mathbf{0}, \sigma^2\mathbf{I}_D)$ (noise)
- $\mathbf{B} \in \mathbb{R}^{D \times M}$ (loading matrix)
- $\boldsymbol{\mu} \in \mathbb{R}^D$ (mean vector)

### Maximum Likelihood Estimation:

**Parameters to estimate:** $\{\mathbf{B}, \boldsymbol{\mu}, \sigma^2\}$

**Likelihood function:**
$$p(\mathbf{X}|\mathbf{B}, \boldsymbol{\mu}, \sigma^2) = \prod_{n=1}^N \mathcal{N}(\mathbf{x}_n; \boldsymbol{\mu}, \mathbf{C})$$

where $\mathbf{C} = \mathbf{B}\mathbf{B}^T + \sigma^2\mathbf{I}_D$ is the covariance matrix.

**ML Estimates:**
- $\hat{\boldsymbol{\mu}} = \frac{1}{N}\sum_{n=1}^N \mathbf{x}_n$ (sample mean)
- $\hat{\mathbf{B}} = \mathbf{U}_M(\boldsymbol{\Lambda}_M - \sigma^2\mathbf{I}_M)^{1/2}\mathbf{R}$
- $\hat{\sigma}^2 = \frac{1}{D-M}\sum_{j=M+1}^D \lambda_j$

where $\mathbf{U}_M$ contains the first $M$ eigenvectors, $\boldsymbol{\Lambda}_M$ contains the first $M$ eigenvalues, and $\mathbf{R}$ is an arbitrary orthogonal matrix.

---

## 6. Mathematical Equivalence

All perspectives lead to the same mathematical solution:

### Common Solution:
The optimal projection matrix $\mathbf{B}$ has columns that are eigenvectors of the sample covariance matrix:

$$\mathbf{S} = \frac{1}{N}\sum_{n=1}^N (\mathbf{x}_n - \boldsymbol{\mu})(\mathbf{x}_n - \boldsymbol{\mu})^T$$

### Eigenvalue Decomposition:
$$\mathbf{S} = \mathbf{U}\boldsymbol{\Lambda}\mathbf{U}^T$$

where:
- $\mathbf{U} = [\mathbf{u}_1, \mathbf{u}_2, \ldots, \mathbf{u}_D]$ (eigenvectors)
- $\boldsymbol{\Lambda} = \text{diag}(\lambda_1, \lambda_2, \ldots, \lambda_D)$ (eigenvalues, $\lambda_1 \geq \lambda_2 \geq \ldots \geq \lambda_D$)

### PCA Solution:
$$\mathbf{B} = [\mathbf{u}_1, \mathbf{u}_2, \ldots, \mathbf{u}_M]$$

The first $M$ eigenvectors corresponding to the largest eigenvalues.

---

## 7. Key Properties and Insights

### Projection Matrix:
$$\mathbf{P} = \mathbf{B}\mathbf{B}^T$$

Properties:
- $\mathbf{P}^2 = \mathbf{P}$ (idempotent)
- $\mathbf{P}^T = \mathbf{P}$ (symmetric)
- Projects onto the $M$-dimensional principal subspace

### Variance Preservation:
The total variance captured by the first $M$ principal components:

$$\text{Variance captured} = \sum_{i=1}^M \lambda_i$$

### Reconstruction Error:
The reconstruction error is:

$$\text{Error} = \sum_{i=M+1}^D \lambda_i$$

### Optimality:
PCA provides the optimal linear dimensionality reduction in terms of:
- Maximum variance preservation
- Minimum reconstruction error
- Maximum mutual information (under Gaussian assumptions)

---

## Conclusion

The multiple perspectives of PCA demonstrate its fundamental importance in machine learning and data analysis. Whether viewed through the lens of variance maximization, error minimization, auto-encoding, or information theory, PCA emerges as the optimal solution for linear dimensionality reduction. This mathematical equivalence provides deep insights into the nature of principal component analysis and its connections to broader concepts in machine learning and information theory.

## 8. Maximum Likelihood Estimation in PPCA

### ML Parameter Estimates

For the PPCA model with parameters $\{\mathbf{B}, \boldsymbol{\mu}, \sigma^2\}$, the maximum likelihood estimates are:

**Mean Parameter:**
$\boldsymbol{\mu}_{\text{ML}} = \frac{1}{N}\sum_{n=1}^N \mathbf{x}_n$

**Loading Matrix:**
$\mathbf{B}_{\text{ML}} = \mathbf{T}(\boldsymbol{\Lambda} - \sigma^2\mathbf{I})^{1/2}\mathbf{R}$

**Noise Variance:**
$\sigma^2_{\text{ML}} = \frac{1}{D-M}\sum_{j=M+1}^D \lambda_j$

where:
- $\mathbf{T} \in \mathbb{R}^{D \times M}$ contains $M$ eigenvectors of the data covariance matrix
- $\boldsymbol{\Lambda} = \text{diag}(\lambda_1, \ldots, \lambda_M) \in \mathbb{R}^{M \times M}$ contains the corresponding eigenvalues
- $\mathbf{R} \in \mathbb{R}^{M \times M}$ is an arbitrary orthogonal matrix

### Key Properties:

**Positive Semidefiniteness:** The matrix $(\boldsymbol{\Lambda} - \sigma^2\mathbf{I})$ is guaranteed to be positive semidefinite since the smallest eigenvalue of the data covariance matrix is bounded from below by the noise variance $\sigma^2$.

**Non-uniqueness:** The ML solution $\mathbf{B}_{\text{ML}}$ is unique only up to an arbitrary orthogonal transformation, representing a singular value decomposition structure.

**Noise Interpretation:** The ML estimate for $\sigma^2$ represents the average variance in the orthogonal complement of the principal subspace - the "leftover" variance not captured by the first $M$ principal components.

---

## 9. Connection Between PPCA and Classical PCA

### Noise-Free Limit

In the limit where $\sigma \to 0$, PPCA and classical PCA provide identical solutions.

**Data Covariance Decomposition:**
Since the data covariance matrix $\mathbf{S}$ is symmetric, it can be diagonalized:

$\mathbf{S} = \mathbf{T}\boldsymbol{\Lambda}\mathbf{T}^{-1}$

**PPCA Covariance Structure:**
In the PPCA model, the data covariance matrix from the Gaussian likelihood $p(\mathbf{x}|\mathbf{B}, \boldsymbol{\mu}, \sigma^2)$ is:

$\text{Cov}[\mathbf{X}] = \mathbf{B}\mathbf{B}^T + \sigma^2\mathbf{I}$

**Convergence Result:**
As $\sigma \to 0$:

$\text{Cov}[\mathbf{X}] = \mathbf{T}\boldsymbol{\Lambda}\mathbf{T}^{-1} = \mathbf{B}\mathbf{B}^T$

This implies:
$\mathbf{B} = \mathbf{T}\boldsymbol{\Lambda}^{1/2}\mathbf{R}$

which matches the ML estimate in equation above with $\sigma = 0$.

---

## 10. Dimensionality Selection and Model Selection

### Automatic Dimensionality Selection

**Gavish-Donoho Heuristic:**
If the noise variance $\sigma^2$ can be estimated, discard all singular values smaller than:

$\tau = 4\sigma\sqrt{3D}$

**Cross-Validation Approach:**
Use nested cross-validation to determine optimal intrinsic dimensionality:

$M^* = \arg\min_M \text{CV-Error}(M)$

**Bayesian Model Selection:**
Apply Bayesian information criteria or model comparison techniques to select appropriate $M$.

### Streaming Data and EM Algorithm

For sequential data arrival, use the iterative Expectation-Maximization (EM) algorithm:

**E-Step:** Compute posterior distribution of latent variables
$q(\mathbf{z}_n) = p(\mathbf{z}_n|\mathbf{x}_n, \mathbf{B}^{(t)}, \sigma^{2(t)})$

**M-Step:** Update parameter estimates
$\{\mathbf{B}^{(t+1)}, \sigma^{2(t+1)}\} = \arg\max_{\mathbf{B}, \sigma^2} \mathbb{E}_{q(\mathbf{z})}[\log p(\mathbf{X}, \mathbf{Z}|\mathbf{B}, \sigma^2)]$

---

## 11. Bayesian PCA

### Bayesian Framework

Instead of point estimates, place prior distributions on parameters:

$p(\boldsymbol{\mu}, \mathbf{B}, \sigma^2)$

**Advantages:**
1. Avoids overfitting issues associated with point estimates
2. Automatic dimensionality selection through model comparison
3. Uncertainty quantification in parameter estimation

**Generative Process:**
$p(\mathbf{x}_n|\mathbf{z}_n, \mathbf{B}, \boldsymbol{\mu}, \sigma^2) = \mathcal{N}(\mathbf{x}_n|\mathbf{B}\mathbf{z}_n + \boldsymbol{\mu}, \sigma^2\mathbf{I})$
$p(\mathbf{z}_n) = \mathcal{N}(\mathbf{z}_n|\mathbf{0}, \mathbf{I})$

**Inference:**
Since analytical integration is intractable, use approximate inference methods:
- **MCMC (Markov Chain Monte Carlo)**
- **Variational Inference**

---

## 12. Extensions and Related Methods

### Factor Analysis (FA)

**Model Extension:**
Allow different noise variances for each dimension:

$p(\mathbf{x}_n|\mathbf{z}_n) = \mathcal{N}(\mathbf{x}_n|\mathbf{B}\mathbf{z}_n + \boldsymbol{\mu}, \text{diag}(\sigma_1^2, \ldots, \sigma_D^2))$

**Properties:**
- More flexible than PPCA
- No closed-form ML solution (requires EM algorithm)
- Stationary points are not necessarily global optima
- Scale-invariant but not rotation-invariant

### Independent Component Analysis (ICA)

**Model Modification:**
Change the prior on latent variables to non-Gaussian distributions:

$p(\mathbf{z}_n) = \prod_{m=1}^M p(z_{nm})$

where $p(z_{nm})$ are non-Gaussian (e.g., Laplacian, uniform)

**Applications:**
- **Blind Source Separation**: Separate mixed signals (e.g., cocktail party problem)
- **Feature Learning**: Discover independent features in data

**Key Insight:**
Non-Gaussian priors allow identification of individual source signals, whereas PCA can only identify the subspace.

---

## 13. Nonlinear Extensions

### Deep Auto-encoders

**Architecture:**
- **Encoder**: $\mathbf{z} = f_{\text{enc}}(\mathbf{x}; \boldsymbol{\theta}_{\text{enc}})$ (multilayer neural network)
- **Decoder**: $\tilde{\mathbf{x}} = f_{\text{dec}}(\mathbf{z}; \boldsymbol{\theta}_{\text{dec}})$ (multilayer neural network)

**Connection to PCA:**
When activation functions are identity functions, the deep auto-encoder reduces to linear PCA.

**Training:**
$\min_{\boldsymbol{\theta}} \frac{1}{N}\sum_{n=1}^N \|\mathbf{x}_n - f_{\text{dec}}(f_{\text{enc}}(\mathbf{x}_n; \boldsymbol{\theta}_{\text{enc}}); \boldsymbol{\theta}_{\text{dec}})\|_2^2$

### Kernel PCA

**Kernel Trick:**
Implicitly map data to infinite-dimensional feature space:

$\phi: \mathbb{R}^D \to \mathcal{H}$

**Kernel Function:**
$k(\mathbf{x}_i, \mathbf{x}_j) = \langle\phi(\mathbf{x}_i), \phi(\mathbf{x}_j)\rangle_{\mathcal{H}}$

**Algorithm:**
1. Compute kernel matrix: $\mathbf{K}_{ij} = k(\mathbf{x}_i, \mathbf{x}_j)$
2. Center the kernel matrix
3. Perform eigendecomposition on centered kernel matrix
4. Extract principal components in feature space

### Gaussian Process Latent Variable Model (GP-LVM)

**Model:**
Replace linear mapping with Gaussian process:

$p(\mathbf{x}_n|\mathbf{z}_n, \mathbf{f}) = \mathcal{N}(\mathbf{x}_n|\mathbf{f}(\mathbf{z}_n), \sigma^2\mathbf{I})$

where $\mathbf{f} \sim \mathcal{GP}(\mathbf{0}, k(\cdot, \cdot))$

**Inference:**
- **GP-LVM**: Point estimates of latent variables $\mathbf{Z}$, marginalize out function $\mathbf{f}$
- **Bayesian GP-LVM**: Maintain distributions over both $\mathbf{Z}$ and $\mathbf{f}$

---

## 14. Summary of Linear Dimensionality Reduction Methods

| Method | Prior on $\mathbf{z}$ | Noise Model | Key Property |
|--------|-------------------|-------------|--------------|
| **PCA** | Deterministic | Uniform noise | Variance maximization |
| **PPCA** | $\mathcal{N}(\mathbf{0}, \mathbf{I})$ | $\sigma^2\mathbf{I}$ | Probabilistic framework |
| **Factor Analysis** | $\mathcal{N}(\mathbf{0}, \mathbf{I})$ | $\text{diag}(\sigma_1^2, \ldots, \sigma_D^2)$ | Flexible noise model |
| **ICA** | Non-Gaussian | $\sigma^2\mathbf{I}$ | Source separation |

Modern applications continue to build upon these foundational concepts, with deep learning approaches like variational autoencoders and generative adversarial networks drawing inspiration from the probabilistic and auto-encoder perspectives of PCA.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import warnings
warnings.filterwarnings('ignore')


class PCA:
    """
    Principal Component Analysis implementation from multiple perspectives.
    
    This class implements PCA using different approaches:
    1. Covariance matrix eigendecomposition
    2. SVD-based approach  
    3. Iterative power method
    4. Probabilistic PCA (PPCA)
    """
    
    def __init__(self, n_components=None, method='eigen'):
        """
        Initialize PCA.
        
        Parameters:
        -----------
        n_components : int or None
            Number of principal components to keep
        method : str
            Method to use: 'eigen', 'svd', 'power', 'ppca'
        """
        self.n_components = n_components
        self.method = method
        self.components_ = None
        self.eigenvalues_ = None
        self.mean_ = None
        self.explained_variance_ratio_ = None
        self.noise_variance_ = None  # For PPCA
        
    def fit(self, X):
        """
        Fit PCA to data.
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Training data
        """
        X = np.array(X, dtype=float)
        n_samples, n_features = X.shape
        
        # Center the data
        self.mean_ = np.mean(X, axis=0)
        X_centered = X - self.mean_
        
        # Determine number of components
        if self.n_components is None:
            self.n_components = min(n_samples, n_features)
        
        if self.method == 'eigen':
            self._fit_eigen(X_centered)
        elif self.method == 'svd':
            self._fit_svd(X_centered)
        elif self.method == 'power':
            self._fit_power_method(X_centered)
        elif self.method == 'ppca':
            self._fit_ppca(X_centered)
        else:
            raise ValueError(f"Unknown method: {self.method}")
            
        return self
    
    def _fit_eigen(self, X_centered):
        """Fit using eigendecomposition of covariance matrix."""
        n_samples = X_centered.shape[0]
        
        # Compute covariance matrix
        cov_matrix = np.cov(X_centered.T)
        
        # Eigendecomposition
        eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
        
        # Sort by eigenvalues (descending)
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]
        
        # Select top components
        self.eigenvalues_ = eigenvalues[:self.n_components]
        self.components_ = eigenvectors[:, :self.n_components].T
        
        # Explained variance ratio
        total_variance = np.sum(eigenvalues)
        self.explained_variance_ratio_ = self.eigenvalues_ / total_variance
    
    def _fit_svd(self, X_centered):
        """Fit using Singular Value Decomposition."""
        n_samples = X_centered.shape[0]
        
        # SVD
        U, s, Vt = np.linalg.svd(X_centered, full_matrices=False)
        
        # Convert singular values to eigenvalues
        eigenvalues = (s ** 2) / (n_samples - 1)
        
        # Select top components
        self.eigenvalues_ = eigenvalues[:self.n_components]
        self.components_ = Vt[:self.n_components]
        
        # Explained variance ratio
        total_variance = np.sum(eigenvalues)
        self.explained_variance_ratio_ = self.eigenvalues_ / total_variance
    
    def _fit_power_method(self, X_centered, max_iter=1000, tol=1e-6):
        """Fit using iterative power method."""
        cov_matrix = np.cov(X_centered.T)
        n_features = cov_matrix.shape[0]
        
        components = []
        eigenvalues = []
        
        # Deflation matrix
        A = cov_matrix.copy()
        
        for i in range(self.n_components):
            # Random initialization
            v = np.random.randn(n_features)
            v = v / np.linalg.norm(v)
            
            # Power iteration
            for _ in range(max_iter):
                v_new = A @ v
                eigenval = np.linalg.norm(v_new)
                v_new = v_new / eigenval
                
                if np.abs(np.dot(v, v_new)) > 1 - tol:
                    break
                v = v_new
            
            eigenvalues.append(eigenval)
            components.append(v)
            
            # Deflation: remove current component
            A = A - eigenval * np.outer(v, v)
        
        self.eigenvalues_ = np.array(eigenvalues)
        self.components_ = np.array(components)
        
        # Explained variance ratio
        total_variance = np.trace(cov_matrix)
        self.explained_variance_ratio_ = self.eigenvalues_ / total_variance
    
    def _fit_ppca(self, X_centered, max_iter=100, tol=1e-6):
        """Fit using Probabilistic PCA with EM algorithm."""
        n_samples, n_features = X_centered.shape
        
        # Initialize parameters
        W = np.random.randn(n_features, self.n_components)
        sigma2 = 1.0
        
        for iteration in range(max_iter):
            # E-step: Compute posterior statistics
            M = W.T @ W + sigma2 * np.eye(self.n_components)
            M_inv = np.linalg.inv(M)
            
            # Posterior mean for each data point
            Ez = X_centered @ W @ M_inv  # (n_samples, n_components)
            
            # Posterior covariance (same for all data points)
            Ezz = sigma2 * M_inv + Ez.T @ Ez / n_samples
            
            # M-step: Update parameters
            W_new = (X_centered.T @ Ez) @ np.linalg.inv(Ezz)
            
            # Update noise variance
            sigma2_new = np.sum((X_centered - Ez @ W_new.T) ** 2) / (n_samples * n_features)
            
            # Check convergence
            if np.linalg.norm(W_new - W) < tol and abs(sigma2_new - sigma2) < tol:
                break
                
            W, sigma2 = W_new, sigma2_new
        
        # Convert to PCA format
        U, s, _ = np.linalg.svd(W, full_matrices=False)
        self.components_ = (U @ np.diag(s)).T
        self.eigenvalues_ = s ** 2 + sigma2
        self.noise_variance_ = sigma2
        
        # Explained variance ratio
        total_variance = np.var(X_centered, axis=0).sum()
        self.explained_variance_ratio_ = self.eigenvalues_ / total_variance
    
    def transform(self, X):
        """
        Transform data to lower dimensional space.
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Data to transform
            
        Returns:
        --------
        X_transformed : array, shape (n_samples, n_components)
            Transformed data
        """
        if self.components_ is None:
            raise ValueError("PCA must be fitted before transform")
            
        X = np.array(X, dtype=float)
        X_centered = X - self.mean_
        return X_centered @ self.components_.T
    
    def inverse_transform(self, X_transformed):
        """
        Transform data back to original space.
        
        Parameters:
        -----------
        X_transformed : array-like, shape (n_samples, n_components)
            Transformed data
            
        Returns:
        --------
        X_reconstructed : array, shape (n_samples, n_features)
            Reconstructed data
        """
        if self.components_ is None:
            raise ValueError("PCA must be fitted before inverse_transform")
            
        X_transformed = np.array(X_transformed, dtype=float)
        return X_transformed @ self.components_ + self.mean_
    
    def fit_transform(self, X):
        """Fit PCA and transform data."""
        return self.fit(X).transform(X)
    
    def reconstruction_error(self, X):
        """Compute reconstruction error."""
        X_transformed = self.transform(X)
        X_reconstructed = self.inverse_transform(X_transformed)
        return np.mean(np.sum((X - X_reconstructed) ** 2, axis=1))


class LinearAutoEncoder:
    """
    Linear Auto-encoder implementation equivalent to PCA.
    """
    
    def __init__(self, n_components, learning_rate=0.01, max_iter=1000):
        self.n_components = n_components
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.encoder_weights = None
        self.decoder_weights = None
        self.mean_ = None
        
    def fit(self, X):
        """Fit the linear auto-encoder."""
        X = np.array(X, dtype=float)
        n_samples, n_features = X.shape
        
        # Center the data
        self.mean_ = np.mean(X, axis=0)
        X_centered = X - self.mean_
        
        # Initialize weights
        self.encoder_weights = np.random.randn(n_features, self.n_components) * 0.01
        self.decoder_weights = np.random.randn(self.n_components, n_features) * 0.01
        
        # Training loop
        for iteration in range(self.max_iter):
            # Forward pass
            Z = X_centered @ self.encoder_weights  # Encoding
            X_reconstructed = Z @ self.decoder_weights  # Decoding
            
            # Compute loss (MSE)
            loss = np.mean((X_centered - X_reconstructed) ** 2)
            
            # Backward pass
            dX = 2 * (X_reconstructed - X_centered) / n_samples
            
            # Gradients
            dW_decoder = Z.T @ dX
            dW_encoder = X_centered.T @ (dX @ self.decoder_weights.T)
            
            # Update weights
            self.decoder_weights -= self.learning_rate * dW_decoder
            self.encoder_weights -= self.learning_rate * dW_encoder
            
            if iteration % 100 == 0:
                print(f"Iteration {iteration}, Loss: {loss:.6f}")
        
        return self
    
    def encode(self, X):
        """Encode data to latent space."""
        X = np.array(X, dtype=float)
        X_centered = X - self.mean_
        return X_centered @ self.encoder_weights
    
    def decode(self, Z):
        """Decode from latent space."""
        Z = np.array(Z, dtype=float)
        return Z @ self.decoder_weights + self.mean_


class KernelPCA:
    """
    Kernel PCA implementation for nonlinear dimensionality reduction.
    """
    
    def __init__(self, n_components, kernel='rbf', gamma=1.0, degree=3, coef0=1):
        self.n_components = n_components
        self.kernel = kernel
        self.gamma = gamma
        self.degree = degree
        self.coef0 = coef0
        self.X_fit_ = None
        self.alphas_ = None
        self.lambdas_ = None
        
    def _kernel_function(self, X1, X2):
        """Compute kernel matrix."""
        if self.kernel == 'linear':
            return X1 @ X2.T
        elif self.kernel == 'poly':
            return (self.gamma * X1 @ X2.T + self.coef0) ** self.degree
        elif self.kernel == 'rbf':
            # RBF kernel
            X1_norm = np.sum(X1 ** 2, axis=1, keepdims=True)
            X2_norm = np.sum(X2 ** 2, axis=1, keepdims=True)
            K = X1_norm + X2_norm.T - 2 * X1 @ X2.T
            return np.exp(-self.gamma * K)
        else:
            raise ValueError(f"Unknown kernel: {self.kernel}")
    
    def fit(self, X):
        """Fit Kernel PCA."""
        X = np.array(X, dtype=float)
        n_samples = X.shape[0]
        self.X_fit_ = X
        
        # Compute kernel matrix
        K = self._kernel_function(X, X)
        
        # Center the kernel matrix
        ones = np.ones((n_samples, n_samples)) / n_samples
        K_centered = K - ones @ K - K @ ones + ones @ K @ ones
        
        # Eigendecomposition
        eigenvalues, eigenvectors = np.linalg.eigh(K_centered)
        
        # Sort by eigenvalues (descending)
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]
        
        # Select top components and normalize
        self.lambdas_ = eigenvalues[:self.n_components]
        self.alphas_ = eigenvectors[:, :self.n_components]
        
        # Normalize eigenvectors
        for i in range(self.n_components):
            self.alphas_[:, i] = self.alphas_[:, i] / np.sqrt(self.lambdas_[i])
        
        return self
    
    def transform(self, X):
        """Transform data using fitted Kernel PCA."""
        X = np.array(X, dtype=float)
        
        # Compute kernel between new data and training data
        K = self._kernel_function(X, self.X_fit_)
        
        # Center the kernel matrix
        n_train = self.X_fit_.shape[0]
        ones_train = np.ones((1, n_train)) / n_train
        ones_test = np.ones((X.shape[0], 1)) / n_train
        
        K_centered = K - ones_test @ ones_train @ self.X_fit_ - K @ ones_train.T + ones_test @ ones_train @ self.X_fit_
        
        # Transform
        return K_centered @ self.alphas_


def generate_sample_data():
    """Generate sample data for testing PCA implementations."""
    np.random.seed(42)
    
    # Generate 2D data with correlation
    mean = [0, 0]
    cov = [[2, 1], [1, 1]]
    data_2d = np.random.multivariate_normal(mean, cov, 200)
    
    # Generate 3D data (2D data embedded in 3D with noise)
    data_3d = np.column_stack([
        data_2d[:, 0] + 0.1 * np.random.randn(200),
        data_2d[:, 1] + 0.1 * np.random.randn(200),
        0.5 * data_2d[:, 0] + 0.3 * data_2d[:, 1] + 0.2 * np.random.randn(200)
    ])
    
    return data_2d, data_3d


def compare_pca_methods():
    """Compare different PCA implementation methods."""
    print("=== Comparing PCA Methods ===\n")
    
    # Generate test data
    _, data_3d = generate_sample_data()
    
    methods = ['eigen', 'svd', 'power', 'ppca']
    results = {}
    
    for method in methods:
        print(f"Method: {method}")
        pca = PCA(n_components=2, method=method)
        pca.fit(data_3d)
        
        print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
        print(f"Reconstruction error: {pca.reconstruction_error(data_3d):.6f}")
        
        results[method] = {
            'components': pca.components_,
            'eigenvalues': pca.eigenvalues_,
            'explained_variance_ratio': pca.explained_variance_ratio_
        }
        print()
    
    return results


def demonstrate_linear_autoencoder():
    """Demonstrate linear auto-encoder equivalence to PCA."""
    print("=== Linear Auto-encoder vs PCA ===\n")
    
    _, data_3d = generate_sample_data()
    
    # PCA
    pca = PCA(n_components=2, method='eigen')
    pca.fit(data_3d)
    pca_transformed = pca.transform(data_3d)
    pca_error = pca.reconstruction_error(data_3d)
    
    # Linear Auto-encoder
    autoencoder = LinearAutoEncoder(n_components=2, learning_rate=0.1, max_iter=1000)
    autoencoder.fit(data_3d)
    ae_encoded = autoencoder.encode(data_3d)
    ae_decoded = autoencoder.decode(ae_encoded)
    ae_error = np.mean(np.sum((data_3d - ae_decoded) ** 2, axis=1))
    
    print(f"PCA reconstruction error: {pca_error:.6f}")
    print(f"Auto-encoder reconstruction error: {ae_error:.6f}")
    print(f"Difference: {abs(pca_error - ae_error):.6f}")


def demonstrate_kernel_pca():
    """Demonstrate Kernel PCA on nonlinear data."""
    print("\n=== Kernel PCA Demo ===\n")
    
    # Generate nonlinear data (circle)
    np.random.seed(42)
    theta = np.linspace(0, 2*np.pi, 100)
    radius = 1 + 0.1 * np.random.randn(100)
    X_circle = np.column_stack([
        radius * np.cos(theta),
        radius * np.sin(theta)
    ])
    
    # Linear PCA (should not work well)
    pca_linear = PCA(n_components=1, method='eigen')
    pca_linear.fit(X_circle)
    print(f"Linear PCA explained variance: {pca_linear.explained_variance_ratio_[0]:.3f}")
    
    # Kernel PCA with RBF kernel
    kpca = KernelPCA(n_components=1, kernel='rbf', gamma=1.0)
    kpca.fit(X_circle)
    X_kpca = kpca.transform(X_circle)
    
    print("Kernel PCA successfully applied to nonlinear data")
    print(f"Transformed data shape: {X_kpca.shape}")


def plot_pca_results():
    """Plot PCA results for visualization."""
    try:
        data_2d, data_3d = generate_sample_data()
        
        # Apply PCA
        pca = PCA(n_components=2, method='eigen')
        pca.fit(data_3d)
        data_transformed = pca.transform(data_3d)
        data_reconstructed = pca.inverse_transform(data_transformed)
        
        # Create plots
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        
        # Original 3D data (first two dimensions)
        axes[0, 0].scatter(data_3d[:, 0], data_3d[:, 1], alpha=0.6)
        axes[0, 0].set_title('Original Data (First 2 Dims)')
        axes[0, 0].set_xlabel('Feature 1')
        axes[0, 0].set_ylabel('Feature 2')
        
        # PCA transformed data
        axes[0, 1].scatter(data_transformed[:, 0], data_transformed[:, 1], alpha=0.6)
        axes[0, 1].set_title('PCA Transformed Data')
        axes[0, 1].set_xlabel('PC 1')
        axes[0, 1].set_ylabel('PC 2')
        
        # Reconstructed data
        axes[1, 0].scatter(data_reconstructed[:, 0], data_reconstructed[:, 1], alpha=0.6)
        axes[1, 0].set_title('Reconstructed Data (First 2 Dims)')
        axes[1, 0].set_xlabel('Feature 1')
        axes[1, 0].set_ylabel('Feature 2')
        
        # Explained variance
        axes[1, 1].bar(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_)
        axes[1, 1].set_title('Explained Variance Ratio')
        axes[1, 1].set_xlabel('Principal Component')
        axes[1, 1].set_ylabel('Explained Variance Ratio')
        
        plt.tight_layout()
        plt.show()
        
    except ImportError:
        print("Matplotlib not available for plotting")


if __name__ == "__main__":
    # Run demonstrations
    compare_pca_methods()
    demonstrate_linear_autoencoder()
    demonstrate_kernel_pca()
    
    # Plot results if matplotlib is available
    plot_pca_results()
    
    print("\n=== All demonstrations completed ===")
```

```python
import math
import random
import matplotlib.pyplot as plt

class PCA:
    """
    Principal Component Analysis implementation with focus on Probabilistic PCA (PPCA).
    """
    
    def __init__(self, n_components=None, method='ppca'):
        """
        Initialize PCA.
        
        Parameters:
        -----------
        n_components : int or None
            Number of principal components to keep
        method : str
            Method to use: 'ppca' (only method implemented here)
        """
        self.n_components = n_components
        self.method = method
        self.components_ = None
        self.eigenvalues_ = None
        self.mean_ = None
        self.explained_variance_ratio_ = None
        self.noise_variance_ = None  # For PPCA
        self.W_ = None  # For PPCA (equivalent to B in the text)
    
    def fit(self, X):
        """
        Fit PCA to data.
        
        Parameters:
        -----------
        X : list of lists, shape (n_samples, n_features)
            Training data
        """
        n_samples = len(X)
        n_features = len(X[0])
        
        # Center the data
        self.mean_ = [0.0] * n_features
        for x in X:
            for d in range(n_features):
                self.mean_[d] += x[d] / n_samples
        X_centered = [[x[d] - self.mean_[d] for d in range(n_features)] for x in X]
        
        # Determine number of components
        if self.n_components is None:
            self.n_components = min(n_samples, n_features)
        
        if self.method == 'ppca':
            self._fit_ppca(X_centered)
        else:
            raise ValueError(f"Only 'ppca' method is implemented in core Python")
        
        return self
    
    def _fit_ppca(self, X_centered, max_iter=100, tol=1e-6):
        """Fit using Probabilistic PCA with EM algorithm (Section 10.7)."""
        n_samples = len(X_centered)
        n_features = len(X_centered[0])
        M = self.n_components
        
        # Initialize parameters (W is equivalent to B, sigma2 is noise variance)
        W = [[random.random() * 0.1 for _ in range(M)] for _ in range(n_features)]
        sigma2 = 1.0
        
        for iteration in range(max_iter):
            # E-step: Compute posterior statistics (Section 10.7.3)
            # M = W^T W + σ^2 I (M x M)
            WTW = [[0.0 for _ in range(M)] for _ in range(M)]
            for m1 in range(M):
                for m2 in range(M):
                    for d in range(n_features):
                        WTW[m1][m2] += W[d][m1] * W[d][m2]
            M_matrix = [[WTW[m1][m2] + (sigma2 if m1 == m2 else 0.0) for m2 in range(M)] for m1 in range(M)]
            
            # Compute M^-1 (for M=1 or M=2, compute manually)
            if M == 1:
                M_inv = [[1.0 / M_matrix[0][0]]]
            else:  # M=2
                det = M_matrix[0][0] * M_matrix[1][1] - M_matrix[0][1] * M_matrix[1][0]
                M_inv = [
                    [M_matrix[1][1] / det, -M_matrix[0][1] / det],
                    [-M_matrix[1][0] / det, M_matrix[0][0] / det]
                ]
            
            # Posterior mean Ez = X W M^-1 (n_samples x M)
            Ez = [[0.0 for _ in range(M)] for _ in range(n_samples)]
            for n in range(n_samples):
                for m in range(M):
                    for d in range(n_features):
                        for m2 in range(M):
                            Ez[n][m] += X_centered[n][d] * W[d][m2] * M_inv[m2][m]
            
            # Posterior covariance Ezz = σ^2 M^-1 + Ez^T Ez / n_samples (M x M)
            EzT_Ez = [[0.0 for _ in range(M)] for _ in range(M)]
            for m1 in range(M):
                for m2 in range(M):
                    for n in range(n_samples):
                        EzT_Ez[m1][m2] += Ez[n][m1] * Ez[n][m2] / n_samples
            Ezz = [[sigma2 * M_inv[m1][m2] + EzT_Ez[m1][m2] for m2 in range(M)] for m1 in range(M)]
            
            # M-step: Update parameters
            # W_new = (X^T Ez) (Ezz)^-1 (D x M)
            XT_Ez = [[0.0 for _ in range(M)] for _ in range(n_features)]
            for d in range(n_features):
                for m in range(M):
                    for n in range(n_samples):
                        XT_Ez[d][m] += X_centered[n][d] * Ez[n][m]
            
            # Compute Ezz^-1
            if M == 1:
                Ezz_inv = [[1.0 / Ezz[0][0]]]
            else:
                det = Ezz[0][0] * Ezz[1][1] - Ezz[0][1] * Ezz[1][0]
                Ezz_inv = [
                    [Ezz[1][1] / det, -Ezz[0][1] / det],
                    [-Ezz[1][0] / det, Ezz[0][0] / det]
                ]
            
            # W_new = (X^T Ez) (Ezz)^-1
            W_new = [[0.0 for _ in range(M)] for _ in range(n_features)]
            for d in range(n_features):
                for m in range(M):
                    for m2 in range(M):
                        W_new[d][m] += XT_Ez[d][m2] * Ezz_inv[m2][m]
            
            # Update noise variance
            Ez_WT = [[0.0 for _ in range(n_features)] for _ in range(n_samples)]
            for n in range(n_samples):
                for d in range(n_features):
                    for m in range(M):
                        Ez_WT[n][d] += Ez[n][m] * W_new[d][m]
            sigma2_new = 0.0
            for n in range(n_samples):
                for d in range(n_features):
                    sigma2_new += (X_centered[n][d] - Ez_WT[n][d]) ** 2
            sigma2_new /= (n_samples * n_features)
            
            # Check convergence
            W_diff = 0.0
            for d in range(n_features):
                for m in range(M):
                    W_diff += (W_new[d][m] - W[d][m]) ** 2
            W_diff = math.sqrt(W_diff)
            sigma2_diff = abs(sigma2_new - sigma2)
            
            if W_diff < tol and sigma2_diff < tol:
                break
            
            W, sigma2 = W_new, sigma2_new
        
        self.W_ = W
        self.noise_variance_ = sigma2
        
        # Approximate components and eigenvalues (simplified)
        self.components_ = W  # In a full implementation, we'd use SVD on W
        self.eigenvalues_ = [1.0] * M  # Placeholder (requires SVD)
        total_variance = sum(sum(x[d] ** 2 for d in range(n_features)) for x in X_centered) / n_samples
        self.explained_variance_ratio_ = [1.0 / M] * M  # Placeholder
    
    def transform(self, X):
        """
        Transform data to lower dimensional space.
        
        Parameters:
        -----------
        X : list of lists, shape (n_samples, n_features)
            Data to transform
            
        Returns:
        --------
        X_transformed : list of lists, shape (n_samples, n_components)
            Transformed data
        """
        if self.components_ is None:
            raise ValueError("PCA must be fitted before transform")
        
        n_features = len(X[0])
        M = self.n_components
        X_centered = [[x[d] - self.mean_[d] for d in range(n_features)] for x in X]
        
        # For PPCA, transform using posterior mean (Section 10.7.3)
        # M = W^T W + σ^2 I
        WTW = [[0.0 for _ in range(M)] for _ in range(M)]
        for m1 in range(M):
            for m2 in range(M):
                for d in range(n_features):
                    WTW[m1][m2] += self.W_[d][m1] * self.W_[d][m2]
        M_matrix = [[WTW[m1][m2] + (self.noise_variance_ if m1 == m2 else 0.0) for m2 in range(M)] for m1 in range(M)]
        
        if M == 1:
            M_inv = [[1.0 / M_matrix[0][0]]]
        else:
            det = M_matrix[0][0] * M_matrix[1][1] - M_matrix[0][1] * M_matrix[1][0]
            M_inv = [
                [M_matrix[1][1] / det, -M_matrix[0][1] / det],
                [-M_matrix[1][0] / det, M_matrix[0][0] / det]
            ]
        
        X_transformed = [[0.0 for _ in range(M)] for _ in range(len(X))]
        for n in range(len(X)):
            for m in range(M):
                for d in range(n_features):
                    for m2 in range(M):
                        X_transformed[n][m] += X_centered[n][d] * self.W_[d][m2] * M_inv[m2][m]
        return X_transformed
    
    def inverse_transform(self, X_transformed):
        """
        Transform data back to original space.
        
        Parameters:
        -----------
        X_transformed : list of lists, shape (n_samples, n_components)
            Transformed data
            
        Returns:
        --------
        X_reconstructed : list of lists, shape (n_samples, n_features)
            Reconstructed data
        """
        if self.components_ is None:
            raise ValueError("PCA must be fitted before inverse_transform")
        
        n_features = len(self.mean_)
        X_reconstructed = [[0.0 for _ in range(n_features)] for _ in range(len(X_transformed))]
        for n in range(len(X_transformed)):
            for d in range(n_features):
                for m in range(self.n_components):
                    X_reconstructed[n][d] += X_transformed[n][m] * self.components_[d][m]
                X_reconstructed[n][d] += self.mean_[d]
        return X_reconstructed
    
    def reconstruction_error(self, X):
        """Compute reconstruction error."""
        X_transformed = self.transform(X)
        X_reconstructed = self.inverse_transform(X_transformed)
        error = 0.0
        for n in range(len(X)):
            for d in range(len(X[0])):
                error += (X[n][d] - X_reconstructed[n][d]) ** 2
        return error / len(X)

def generate_sample_data():
    """Generate sample 2D data with correlation (equivalent to original)."""
    random.seed(42)
    
    # Simulate 2D data with correlation (mean=[0,0], cov=[[2,1],[1,1]])
    data_2d = []
    for _ in range(200):
        # Generate correlated data using a simplified method
        z1 = random.gauss(0, math.sqrt(2))  # Variance 2
        z2 = (z1 * 0.5) + random.gauss(0, math.sqrt(0.75))  # Covariance 1, variance 1
        data_2d.append([z1, z2])
    
    return data_2d

def plot_pca_results():
    """Plot PCA results for visualization."""
    data_2d = generate_sample_data()
    
    # Apply PCA
    pca = PCA(n_components=1, method='ppca')
    pca.fit(data_2d)
    data_transformed = pca.transform(data_2d)
    data_reconstructed = pca.inverse_transform(data_transformed)
    
    # Create plots
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    # Original 2D data
    axes[0, 0].scatter([x[0] for x in data_2d], [x[1] for x in data_2d], alpha=0.6)
    axes[0, 0].set_title('Original Data (2D)')
    axes[0, 0].set_xlabel('Feature 1')
    axes[0, 0].set_ylabel('Feature 2')
    axes[0, 0].axis('equal')
    
    # PCA transformed data (1D)
    axes[0, 1].scatter([x[0] for x in data_transformed], [0] * len(data_transformed), alpha=0.6)
    axes[0, 1].set_title('PPCA Transformed Data (1D)')
    axes[0, 1].set_xlabel('Latent Variable z')
    axes[0, 1].set_yticks([])
    
    # Reconstructed data
    axes[1, 0].scatter([x[0] for x in data_reconstructed], [x[1] for x in data_reconstructed], alpha=0.6)
    axes[1, 0].set_title('Reconstructed Data (2D)')
    axes[1, 0].set_xlabel('Feature 1')
    axes[1, 0].set_ylabel('Feature 2')
    axes[1, 0].axis('equal')
    
    # Explained variance (placeholder)
    axes[1, 1].bar(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_)
    axes[1, 1].set_title('Explained Variance Ratio (Placeholder)')
    axes[1, 1].set_xlabel('Principal Component')
    axes[1, 1].set_ylabel('Explained Variance Ratio')
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("PPCA Implementation in Core Python")
    print("=" * 60)
    
    # Generate and plot results
    plot_pca_results()
    
    print("\n=== Demonstration completed ===")
```
