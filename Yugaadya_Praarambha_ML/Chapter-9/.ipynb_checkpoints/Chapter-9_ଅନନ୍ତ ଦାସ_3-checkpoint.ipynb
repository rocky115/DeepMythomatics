## Latent Variable Perspective

In the previous sections, we derived PCA using the maximum-variance and projection perspectives, without relying on a probabilistic model. While this deterministic approach simplifies the mathematics by avoiding the complexities of probability theory, it limits our ability to leverage the benefits that a probabilistic framework offers. A probabilistic model for PCA provides greater flexibility and valuable insights, including the ability to:

- Define a likelihood function, enabling maximum likelihood estimation.
- Explicitly handle noisy observations, which were not addressed in earlier deterministic methods.
- Perform Bayesian model comparison using the marginal likelihood, as discussed in Section 8.6.
- View PCA as a generative model, allowing us to simulate new data points.
- Establish connections to related algorithms, such as factor analysis.
- Address missing data (missing at random) by applying Bayes’ theorem.
- Quantify the novelty of a new data point (e.g., for anomaly detection).
- Extend the model in a principled way, such as creating a mixture of PCA models.
- Recover the deterministic PCA (from earlier sections) as a special case in a noise-free setting.
- Enable a fully Bayesian treatment by marginalizing out model parameters.

To achieve these benefits, we introduce a continuous-valued latent variable $ z \in \mathbb{R}^M $, where $ M $ is the dimensionality of the latent space (typically $ M < D $). Using this latent variable, PCA can be reformulated as a probabilistic latent-variable model. Tipping and Bishop (1999) proposed this approach, known as **Probabilistic PCA (PPCA)**. PPCA addresses most of the aforementioned limitations, and the deterministic PCA solution—derived by maximizing variance or minimizing reconstruction error—emerges as a special case of maximum likelihood estimation in a noise-free setting.

## Generative Process and Probabilistic Model

In Probabilistic PCA (PPCA), we explicitly define a probabilistic model for linear dimensionality reduction. The model assumes a continuous latent variable $ z \in \mathbb{R}^M $ with a standard normal prior:

$$
p(z) = \mathcal{N}(z \mid 0, I), \tag{10.65}
$$

where $ I $ is the $ M \times M $ identity matrix, meaning $ z $ has mean 0 and covariance $ I $.

The relationship between the latent variable $ z $ and the observed data $ x \in \mathbb{R}^D $ is modeled as a linear mapping with additive Gaussian noise:

$$
x = B z + \mu + \epsilon, \quad \epsilon \sim \mathcal{N}(0, \sigma^2 I), \tag{10.63}
$$

where:
- $ B \in \mathbb{R}^{D \times M} $ is a matrix that maps the latent variables to the observed space,
- $ \mu \in \mathbb{R}^D $ is the mean of the observed data,
- $ \epsilon \in \mathbb{R}^D $ is the Gaussian observation noise with covariance $ \sigma^2 I $, where $ I $ is the $ D \times D $ identity matrix.

This linear relationship implies that the conditional distribution of the observed data given the latent variable is also Gaussian:

$$
p(x \mid z, B, \mu, \sigma^2) = \mathcal{N}(x \mid B z + \mu, \sigma^2 I). \tag{10.64}
$$

### Generative Process

PPCA defines the following generative process to produce observed data points:
1. Sample a latent variable $ z_n $ from the prior:

$$
z_n \sim \mathcal{N}(z \mid 0, I), \tag{10.65}
$$

2. Given $ z_n $, sample an observed data point $ x_n $:

$$
x_n \mid z_n \sim \mathcal{N}(x \mid B z_n + \mu, \sigma^2 I). \tag{10.66}
$$

To generate a data point typical of the model, we use **ancestral sampling**:
- First, sample $ z_n \sim p(z) $ from the standard normal prior.
- Then, use $ z_n $ to sample $ x_n \sim p(x \mid z_n, B, \mu, \sigma^2) $ according to Equation (10.64).

### Joint Distribution and Graphical Model

The generative process allows us to write the joint distribution of the observed data $ x $ and latent variable $ z $ (as discussed in Section 8.4):

$$
p(x, z \mid B, \mu, \sigma^2) = p(x \mid z, B, \mu, \sigma^2) p(z), \tag{10.67}
$$

where $ p(x \mid z, B, \mu, \sigma^2) $ is given by Equation (10.64), and $ p(z) $ is given by Equation (10.65). This joint distribution corresponds to a graphical model, as described in Section 8.5, which is illustrated in Figure 10.14. In this graphical model:
- $ z $ is a parent node (latent variable),
- $ x $ is a child node (observed variable),
- The parameters $ B $, $ \mu $, and $ \sigma^2 $ govern the conditional relationship between $ z $ and $ x $.

### Graphical Model for Probabilistic PCA

Figure 10.14 illustrates the graphical model for Probabilistic PCA (PPCA). In this model:
- The latent variables $ z_n \in \mathbb{R}^M $ are drawn from a standard normal distribution:

$$
z_n \sim \mathcal{N}(0, I), \tag{10.65}
$$

- The observations $ x_n \in \mathbb{R}^D $ explicitly depend on their corresponding latent variables $ z_n $ through the conditional distribution:

$$
p(x_n \mid z_n, B, \mu, \sigma^2) = \mathcal{N}(x_n \mid B z_n + \mu, \sigma^2 I), \tag{10.66}
$$

- The model parameters $ B \in \mathbb{R}^{D \times M} $, $ \mu \in \mathbb{R}^D $, and the likelihood parameter $ \sigma^2 $ are shared across the dataset, meaning they are the same for all $ n = 1, \ldots, N $ data points.

**Remark**: Consider the direction of the arrow in the graphical model connecting the latent variables $ z_n $ to the observed data $ x_n $: the arrow points from $ z_n $ to $ x_n $. This indicates that PPCA assumes a lower-dimensional latent cause $ z_n $ generates the high-dimensional observations $ x_n $. However, in practice, we are often interested in the inverse problem—determining the latent variables $ z_n $ given the observations $ x_n $. To achieve this, we apply Bayesian inference to "invert" the arrow implicitly, inferring the latent variables from the observed data. $ \diamond $

## Example 10.5: Generating New Data Using Latent Variables

PPCA’s generative framework allows us to synthesize new data points by sampling from the latent space. Given the model parameters $ B $ and $ \mu $, we can generate a new data point $ \tilde{x} $ by sampling a latent variable $ z $ and applying the linear mapping:

$$
\tilde{x} = B z + \mu.
$$

If $ \sigma^2 > 0 $, we can also add Gaussian noise $ \epsilon \sim \mathcal{N}(0, \sigma^2 I) $ to simulate observation noise, as in Equation (10.63). The closer the sampled $ z $ is to the latent representations of the training data, the more realistic the generated data $ \tilde{x} $ will be.

### Generating MNIST Digits

Figure 10.15 demonstrates this process using the MNIST digits dataset, focusing on the digit "8". We apply PCA (or PPCA in the noise-free limit) to the subset of MNIST digits "8", using a two-dimensional principal subspace ($ M = 2 $). The latent coordinates $ z_n $ of the training data points are plotted as blue dots in the 2D latent space.

To generate new digits:
- Select a latent vector $ z^* $ in the 2D latent space.
- Compute the corresponding image $ \tilde{x}^* = B z^* + \mu $, where $ B $ is the matrix of principal components (or the mapping matrix in PPCA), and $ \mu $ is the mean of the training data.

Figure 10.15 shows eight generated images of the digit "8", each corresponding to a different $ z^* $ in the latent space. The generated images vary in appearance (e.g., shape, rotation, size) depending on the location of $ z^* $:
- When $ z^* $ is close to the latent coordinates of the training data, the generated digits resemble typical "8"s.
- When $ z^* $ is farther from the training data (e.g., in the top-left and top-right regions of the latent space), the generated digits exhibit more artifacts and look less realistic.

This example highlights that the intrinsic dimensionality of the generated images is only two (since $ M = 2 $), despite the high dimensionality of the observed space ($ D = 784 $ for $ 28 \times 28 $ MNIST images). The 2D latent space captures the primary modes of variation in the data, allowing us to generate new digits by exploring this low-dimensional space.

## Example 10.5: Generating New Data Using Latent Variables

PPCA’s generative framework allows us to synthesize new data points by sampling from the latent space. Given the model parameters $ B $ and $ \mu $, we can generate a new data point $ \tilde{x} $ by sampling a latent variable $ z $ and applying the linear mapping:

$$
\tilde{x} = B z + \mu.
$$

If $ \sigma^2 > 0 $, we can also add Gaussian noise $ \epsilon \sim \mathcal{N}(0, \sigma^2 I) $ to simulate observation noise, as in Equation (10.63). The closer the sampled $ z $ is to the latent representations of the training data, the more realistic the generated data $ \tilde{x} $ will be.

### Generating MNIST Digits

Figure 10.15 demonstrates this process using the MNIST digits dataset, focusing on the digit "8". We apply PCA (or PPCA in the noise-free limit) to the subset of MNIST digits "8", using a two-dimensional principal subspace ($ M = 2 $). The latent coordinates $ z_n $ of the training data points are plotted as blue dots in the 2D latent space.

To generate new digits:
- Select a latent vector $ z^* $ in the 2D latent space.
- Compute the corresponding image $ \tilde{x}^* = B z^* + \mu $, where $ B $ is the matrix of principal components (or the mapping matrix in PPCA), and $ \mu $ is the mean of the training data.

Figure 10.15 shows eight generated images of the digit "8", each corresponding to a different $ z^* $ in the latent space. The generated images vary in appearance (e.g., shape, rotation, size) depending on the location of $ z^* $:
- When $ z^* $ is close to the latent coordinates of the training data, the generated digits resemble typical "8"s.
- When $ z^* $ is farther from the training data (e.g., in the top-left and top-right regions of the latent space), the generated digits exhibit more artifacts and look less realistic.

This example highlights that the intrinsic dimensionality of the generated images is only two (since $ M = 2 $), despite the high dimensionality of the observed space ($ D = 784 $ for $ 28 \times 28 $ MNIST images). The 2D latent space captures the primary modes of variation in the data, allowing us to generate new digits by exploring this low-dimensional space.

```python
import math
import random
import matplotlib.pyplot as plt

# --- Part 1: PPCA on 2D Dataset ---
# Dataset (D = 2, N = 5, same as previous)
data_2d = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N_2d = len(data_2d)
D_2d = len(data_2d[0])
M_2d = 1  # 1D latent space

print("=== Part 1: PPCA on 2D Dataset ===")
print("Original 2D Data:", data_2d)

# Step 1: Compute mean (μ)
def compute_mean(data, D):
    mean = [0.0] * D
    N = len(data)
    for x in data:
        for i in range(D):
            mean[i] += x[i] / N
    return mean

mu_2d = compute_mean(data_2d, D_2d)
print("\nMean (μ):", mu_2d)

# Step 2: Center the data
def center_data(data, mean, D):
    centered_data = []
    for x in data:
        centered_x = [x[i] - mean[i] for i in range(D)]
        centered_data.append(centered_x)
    return centered_data

X_centered_2d = center_data(data_2d, mu_2d, D_2d)
print("Centered Data:", X_centered_2d)

# Step 3: Compute covariance matrix S = (1/N) X X^T (D x D)
X_2d = [[X_centered_2d[n][d] for n in range(N_2d)] for d in range(D_2d)]
XT_2d = [[X_2d[d][n] for d in range(D_2d)] for n in range(N_2d)]

def matrix_multiply(A, B, A_rows, A_cols, B_cols):
    result = [[0.0 for _ in range(B_cols)] for _ in range(A_rows)]
    for i in range(A_rows):
        for j in range(B_cols):
            for k in range(A_cols):
                result[i][j] += A[i][k] * B[k][j]
    return result

S_2d = matrix_multiply(X_2d, XT_2d, D_2d, N_2d, D_2d)
for i in range(D_2d):
    for j in range(D_2d):
        S_2d[i][j] /= N_2d
print("\nCovariance Matrix S:")
for row in S_2d:
    print([f"{val:.3f}" for val in row])

# Step 4: Eigendecomposition of S to estimate B (in noise-free limit, B is related to eigenvectors)
def compute_eigenvalues_2x2(matrix):
    a, b = matrix[0][0], matrix[0][1]
    c, d = matrix[1][0], matrix[1][1]
    trace = a + d
    det = a * d - b * c
    discriminant = trace**2 - 4 * det
    sqrt_disc = math.sqrt(discriminant)
    lambda1 = (trace + sqrt_disc) / 2
    lambda2 = (trace - sqrt_disc) / 2
    return lambda1, lambda2

def compute_eigenvector_2x2(matrix, eigenvalue):
    a, b = matrix[0][0], matrix[0][1]
    c, d = matrix[1][0], matrix[1][1]
    if abs(a - eigenvalue) > 1e-10:
        v1 = -b / (a - eigenvalue)
        v2 = 1.0
    else:
        v1 = 1.0
        v2 = -(d - eigenvalue) / c if abs(c) > 1e-10 else 0.0
    norm = math.sqrt(v1**2 + v2**2)
    return [v1 / norm, v2 / norm]

eigenvalues_2d = compute_eigenvalues_2x2(S_2d)
u1_2d = compute_eigenvector_2x2(S_2d, eigenvalues_2d[0])
u2_2d = compute_eigenvector_2x2(S_2d, eigenvalues_2d[1])
print("\nEigenvalues of S:", eigenvalues_2d)
print("Eigenvectors: u1 =", u1_2d, ", u2 =", u2_2d)

# B is the top M eigenvectors scaled by sqrt(eigenvalue) in PPCA (simplified)
B_2d = [[u1_2d[d]] for d in range(D_2d)]  # D x M matrix (2 x 1)
print("\nEstimated B (D x M):")
for row in B_2d:
    print([f"{val:.3f}" for val in row])

# Estimate σ^2 as the average of the remaining eigenvalues
sigma2_2d = eigenvalues_2d[1]  # Since M=1, use the second eigenvalue
print("Estimated σ^2:", sigma2_2d)

# Step 5: Infer latent variables z_n for each x_n (simplified Bayesian inference)
# In PPCA, p(z|x) ~ N(M^-1 B^T (x - μ), σ^2 M^-1), where M = B^T B + σ^2 I
# For M=1, this simplifies significantly
z_2d = []
for n in range(N_2d):
    x_minus_mu = [X_centered_2d[n][d] for d in range(D_2d)]
    # Compute z_n = (B^T B + σ^2 I)^(-1) B^T (x - μ)
    BTB = 0.0
    for d in range(D_2d):
        BTB += B_2d[d][0] * B_2d[d][0]
    M_inv = 1.0 / (BTB + sigma2_2d)
    BT_x_minus_mu = 0.0
    for d in range(D_2d):
        BT_x_minus_mu += B_2d[d][0] * x_minus_mu[d]
    z_n = [M_inv * BT_x_minus_mu]
    z_2d.append(z_n)
print("\nInferred Latent Variables z_n:")
for n, zn in enumerate(z_2d):
    print(f"z_{n+1}: {zn}")

# Step 6: Generate new data points
def sample_gaussian(mean, variance):
    # Box-Muller transform to sample from N(mean, variance)
    u1, u2 = random.random(), random.random()
    z = math.sqrt(-2.0 * math.log(u1)) * math.cos(2.0 * math.pi * u2)
    return mean + math.sqrt(variance) * z

# Sample new z values
z_new_2d = [[sample_gaussian(0, 1)] for _ in range(3)]  # Generate 3 new samples
print("\nSampled New Latent Variables z_new:")
for i, z in enumerate(z_new_2d):
    print(f"z_new_{i+1}: {z}")

# Generate x_new = B z_new + μ + ϵ
x_new_2d = []
for z in z_new_2d:
    # Compute B z_new
    Bz = [0.0] * D_2d
    for d in range(D_2d):
        for m in range(M_2d):
            Bz[d] += B_2d[d][m] * z[m]
    # Add μ
    x = [Bz[d] + mu_2d[d] for d in range(D_2d)]
    # Add noise ϵ ~ N(0, σ^2 I)
    for d in range(D_2d):
        x[d] += sample_gaussian(0, sigma2_2d)
    x_new_2d.append(x)
print("\nGenerated New Data Points x_new:")
for i, x in enumerate(x_new_2d):
    print(f"x_new_{i+1}: {x}")

# --- Part 2: PPCA on Simulated MNIST-like Dataset ---
# Simulate a small "MNIST-like" dataset for digit "8" (D = 4, N = 5, M = 2)
# Each "image" is a 4D vector (instead of 784D)
data_mnist = [
    [1.0, 2.0, 1.5, 2.5],  # Simulated pixel values for digit "8"
    [1.1, 2.1, 1.6, 2.4],
    [0.9, 1.9, 1.4, 2.6],
    [1.2, 2.2, 1.7, 2.3],
    [1.0, 2.0, 1.5, 2.5]
]
N_mnist = len(data_mnist)
D_mnist = len(data_mnist[0])
M_mnist = 2  # 2D latent space

print("\n=== Part 2: PPCA on Simulated MNIST-like Dataset ===")
print("Simulated MNIST Data (Digit '8'):", data_mnist)

# Step 1: Compute mean (μ)
mu_mnist = compute_mean(data_mnist, D_mnist)
print("\nMean (μ):", mu_mnist)

# Step 2: Center the data
X_centered_mnist = center_data(data_mnist, mu_mnist, D_mnist)
print("Centered Data:", X_centered_mnist)

# Step 3: Compute X^T X (N x N) for high-dimensional approach
X_mnist = [[X_centered_mnist[n][d] for n in range(N_mnist)] for d in range(D_mnist)]
XT_mnist = [[X_mnist[d][n] for d in range(D_mnist)] for n in range(N_mnist)]
XTX_mnist = matrix_multiply(XT_mnist, X_mnist, N_mnist, D_mnist, N_mnist)
for i in range(N_mnist):
    for j in range(N_mnist):
        XTX_mnist[i][j] /= N_mnist
print("\n(1/N) X^T X (N x N):")
for row in XTX_mnist:
    print([f"{val:.3f}" for val in row])

# Step 4: Eigendecomposition of (1/N) X^T X (simplified for N=5)
# Since D=4, N=5, rank is at most 4, so we approximate by using S instead
S_mnist = matrix_multiply(X_mnist, XT_mnist, D_mnist, N_mnist, D_mnist)
for i in range(D_mnist):
    for j in range(D_mnist):
        S_mnist[i][j] /= N_mnist

# For D=4, we implement a simple power iteration to find top 2 eigenvectors
def power_iteration(matrix, D, num_iterations=100):
    # Initialize random vector
    v = [1.0] * D
    norm = math.sqrt(sum(vi**2 for vi in v))
    v = [vi / norm for vi in v]
    # Power iteration
    for _ in range(num_iterations):
        v_new = [0.0] * D
        for i in range(D):
            for j in range(D):
                v_new[i] += matrix[i][j] * v[j]
        norm = math.sqrt(sum(vi**2 for vi in v_new))
        v = [vi / norm if norm > 1e-10 else 0.0 for vi in v_new]
    # Compute eigenvalue
    lambda_val = 0.0
    v_new = [0.0] * D
    for i in range(D):
        for j in range(D):
            v_new[i] += matrix[i][j] * v[j]
        lambda_val += v_new[i] * v[i]
    return lambda_val, v

# Find top 2 eigenvectors
eigenvalues_mnist = []
eigenvectors_mnist = []
# First eigenvector
lambda1, v1 = power_iteration(S_mnist, D_mnist)
eigenvalues_mnist.append(lambda1)
eigenvectors_mnist.append(v1)
# Deflate matrix to find second eigenvector
S_deflated = [[S_mnist[i][j] - lambda1 * v1[i] * v1[j] for j in range(D_mnist)] for i in range(D_mnist)]
lambda2, v2 = power_iteration(S_deflated, D_mnist)
eigenvalues_mnist.append(lambda2)
eigenvectors_mnist.append(v2)

print("\nTop 2 Eigenvalues of S:", eigenvalues_mnist)
print("Top 2 Eigenvectors:")
print("u1 =", eigenvectors_mnist[0])
print("u2 =", eigenvectors_mnist[1])

# B is the top M eigenvectors (D x M)
B_mnist = [[eigenvectors_mnist[m][d] for m in range(M_mnist)] for d in range(D_mnist)]
print("\nEstimated B (D x M):")
for row in B_mnist:
    print([f"{val:.3f}" for val in row])

# Estimate σ^2 as the average of remaining eigenvalues
sigma2_mnist = sum(eigenvalues_mnist[M_mnist:]) / (D_mnist - M_mnist)
print("Estimated σ^2:", sigma2_mnist)

# Step 5: Infer latent variables z_n (2D latent space)
z_mnist = []
for n in range(N_mnist):
    x_minus_mu = [X_centered_mnist[n][d] for d in range(D_mnist)]
    # Compute z_n = (B^T B + σ^2 I)^(-1) B^T (x - μ)
    BTB = [[0.0 for _ in range(M_mnist)] for _ in range(M_mnist)]
    for m1 in range(M_mnist):
        for m2 in range(M_mnist):
            for d in range(D_mnist):
                BTB[m1][m2] += B_mnist[d][m1] * B_mnist[d][m2]
    # Add σ^2 I
    for m in range(M_mnist):
        BTB[m][m] += sigma2_mnist
    # For 2x2 matrix, compute inverse manually
    det = BTB[0][0] * BTB[1][1] - BTB[0][1] * BTB[1][0]
    if abs(det) < 1e-10:
        M_inv = [[1.0 if i == j else 0.0 for j in range(M_mnist)] for i in range(M_mnist)]
    else:
        M_inv = [[BTB[1][1]/det, -BTB[0][1]/det], [-BTB[1][0]/det, BTB[0][0]/det]]
    # Compute B^T (x - μ)
    BT_x_minus_mu = [0.0] * M_mnist
    for m in range(M_mnist):
        for d in range(D_mnist):
            BT_x_minus_mu[m] += B_mnist[d][m] * x_minus_mu[d]
    # z_n = M_inv * B^T (x - μ)
    z_n = [0.0] * M_mnist
    for m1 in range(M_mnist):
        for m2 in range(M_mnist):
            z_n[m1] += M_inv[m1][m2] * BT_x_minus_mu[m2]
    z_mnist.append(z_n)
print("\nInferred Latent Variables z_n (2D):")
for n, zn in enumerate(z_mnist):
    print(f"z_{n+1}: {zn}")

# Step 6: Generate new data points (mimic Figure 10.15)
# Sample 8 new z values in the 2D latent space (e.g., grid points)
z_new_mnist = [
    [-2.0, -2.0], [-2.0, 0.0], [-2.0, 2.0],  # Top row (left to right)
    [0.0, -2.0],  [0.0, 0.0],  [0.0, 2.0],   # Middle row
    [2.0, -2.0],  [2.0, 0.0],  [2.0, 2.0]    # Bottom row
]
print("\nSampled New Latent Variables z_new (Grid Points):")
for i, z in enumerate(z_new_mnist):
    print(f"z_new_{i+1}: {z}")

# Generate x_new = B z_new + μ + ϵ
x_new_mnist = []
for z in z_new_mnist:
    Bz = [0.0] * D_mnist
    for d in range(D_mnist):
        for m in range(M_mnist):
            Bz[d] += B_mnist[d][m] * z[m]
    x = [Bz[d] + mu_mnist[d] for d in range(D_mnist)]
    for d in range(D_mnist):
        x[d] += sample_gaussian(0, sigma2_mnist)
    x_new_mnist.append(x)
print("\nGenerated New Data Points x_new (Simulated Digit '8'):")
for i, x in enumerate(x_new_mnist):
    print(f"x_new_{i+1}: {x}")

# --- Plotting for Part 1 (2D Dataset) ---
def plot_ppca_2d(data, z_2d, x_new_2d, z_new_2d):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    # Plot 1: Observed space with generated points
    data_x = [x[0] for x in data]
    data_y = [x[1] for x in data]
    new_x = [x[0] for x in x_new_2d]
    new_y = [x[1] for x in x_new_2d]
    ax1.scatter(data_x, data_y, color='blue', label='Original Data')
    ax1.scatter(new_x, new_y, color='red', label='Generated Data', marker='x')
    ax1.set_title('2D Data: Original and Generated')
    ax1.set_xlabel('x1')
    ax1.set_ylabel('x2')
    ax1.legend()
    ax1.grid(True)
    ax1.axis('equal')

    # Plot 2: Latent space (1D)
    z_values = [z[0] for z in z_2d]
    z_new_values = [z[0] for z in z_new_2d]
    ax2.scatter(z_values, [0] * len(z_values), color='blue', label='Inferred z_n')
    ax2.scatter(z_new_values, [0] * len(z_new_values), color='red', label='Sampled z_new', marker='x')
    ax2.set_title('1D Latent Space')
    ax2.set_xlabel('z')
    ax2.set_yticks([])
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

# --- Plotting for Part 2 (MNIST-like Dataset) ---
def plot_latent_space_mnist(z_mnist, z_new_mnist):
    fig, ax = plt.subplots(figsize=(6, 6))
    z_x = [z[0] for z in z_mnist]
    z_y = [z[1] for z in z_mnist]
    z_new_x = [z[0] for z in z_new_mnist]
    z_new_y = [z[1] for z in z_new_mnist]
    ax.scatter(z_x, z_y, color='blue', label='Inferred z_n (Training Data)')
    for i, (x, y) in enumerate(zip(z_new_x, z_new_y)):
        ax.scatter(x, y, color='red', marker='x')
        ax.text(x + 0.2, y, f'x_new_{i+1}', color='red')
    ax.set_title('2D Latent Space for Simulated MNIST Digit "8"')
    ax.set_xlabel('z1')
    ax.set_ylabel('z2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PPCA Implementation: Generative Process and Data Generation")
    print("=" * 60)

    # Part 1
    print("\nPlotting Results for 2D Dataset")
    plot_ppca_2d(data_2d, z_2d, x_new_2d, z_new_2d)

    # Part 2
    print("\nPlotting Latent Space for Simulated MNIST Dataset")
    plot_latent_space_mnist(z_mnist, z_new_mnist)

    print("\n" + "=" * 60)
    print("Summary of Results:")
    print("• Implemented PPCA on a 2D dataset and a simulated MNIST-like dataset.")
    print("• Generated new data points using the generative process.")
    print("• Visualized the latent space and generated samples.")
```

## 10.7.3 Posterior Distribution

The joint Gaussian distribution $ p(x, z \mid B, \mu, \sigma^2) $ derived in Equation (10.72) enables us to compute the posterior distribution $ p(z \mid x) $ of the latent variables given the observed data $ x $. Using the rules of Gaussian conditioning from Section 6.5.1, the posterior distribution is also Gaussian:

$$
p(z \mid x) = \mathcal{N}(z \mid m, C), \tag{10.73}
$$

where the mean $ m $ and covariance $ C $ are given by:

$$
m = B^\top (B B^\top + \sigma^2 I)^{-1} (x - \mu), \tag{10.74}
$$

$$
C = I - B^\top (B B^\top + \sigma^2 I)^{-1} B. \tag{10.75}
$$

### Interpretation

- **Posterior Mean**: The mean $ m $ in Equation (10.74) represents the expected latent variable \( z \) given the observation $ x $, effectively mapping \( x \) back to the latent space.
- **Posterior Covariance**: The covariance \( C \) in Equation (10.75) does not depend on the observed data \( x \), meaning the uncertainty in the latent variables is determined solely by the model parameters \( B \) and \( \sigma^2 \).

For a new observation $ x^* $, we can use Equation (10.73) to compute the posterior distribution $ p(z^* \mid x^*) $, which describes the distribution of the corresponding latent variable $ z^* $. The covariance matrix \( C \) provides insight into the confidence of the latent embedding:
- A small determinant of \( C \) (which measures the volume of the covariance ellipsoid) indicates that the latent embedding $ z^* $ is estimated with high certainty.
- A large determinant, or high variance in $ p(z^* \mid x^*) $, suggests that $ x^* $ may be an outlier, as the model is less certain about its latent representation.

### Exploring the Posterior Distribution

We can explore the posterior distribution $ p(z^* \mid x^*) $ to understand what other data points \( x \) are plausible under this posterior. This is done by leveraging the generative process of PPCA (from Section 10.7.1) in the following steps:

1. Sample a latent variable $ z^* \sim p(z \mid x^*) $ from the posterior distribution in Equation (10.73).
2. Sample a reconstructed data point $ \tilde{x}^* \sim p(x \mid z^*, B, \mu, \sigma^2) $, where the conditional distribution is:

$$
p(x \mid z^*, B, \mu, \sigma^2) = \mathcal{N}(x \mid B z^* + \mu, \sigma^2 I), \tag{10.64}
$$

By repeating this sampling process multiple times, we can explore the posterior distribution $ p(z^* \mid x^*) $ and its implications on the observed data space. This sampling process effectively generates hypothetical data points that are plausible under the posterior distribution, allowing us to assess the variability and uncertainty associated with the latent embedding of $ x^* $.

```python
import math
import random
import matplotlib.pyplot as plt

# --- Part 1: Posterior Distribution on 2D Dataset ---
# Dataset (D = 2, N = 5, M = 1, same as previous)
data_2d = [
    [1.0, 2.0],
    [2.0, 1.5],
    [3.0, 3.0],
    [4.0, 3.5],
    [5.0, 4.0]
]
N_2d = len(data_2d)
D_2d = len(data_2d[0])
M_2d = 1

# New observation x* (same as previous for consistency)
x_star_2d = [2.5, 2.5]

# Model parameters from prior implementation
mu_2d = [3.0, 2.8]
B_2d = [[0.906179845938664], [0.4228854651552394]]  # D x M
sigma2_2d = 0.16557280900008452

print("=== Part 1: Posterior Distribution on 2D Dataset ===")
print("New Observation x*:", x_star_2d)

# Step 1: Compute Posterior p(z|x*) = N(z|m, C)
# m = B^T (BB^T + σ^2 I)^(-1) (x - μ)
# C = I - B^T (BB^T + σ^2 I)^(-1) B

# Compute x - μ
x_minus_mu_2d = [x_star_2d[d] - mu_2d[d] for d in range(D_2d)]

# Compute BB^T (D x D)
BBT_2d = [[0.0 for _ in range(D_2d)] for _ in range(D_2d)]
for d1 in range(D_2d):
    for d2 in range(D_2d):
        for m in range(M_2d):
            BBT_2d[d1][d2] += B_2d[d1][m] * B_2d[d2][m]

# Compute BB^T + σ^2 I
BBT_plus_sigma2I_2d = [[BBT_2d[d1][d2] + (sigma2_2d if d1 == d2 else 0.0) for d2 in range(D_2d)] for d1 in range(D_2d)]

# Compute (BB^T + σ^2 I)^(-1) for 2x2 matrix
a, b = BBT_plus_sigma2I_2d[0][0], BBT_plus_sigma2I_2d[0][1]
c, d = BBT_plus_sigma2I_2d[1][0], BBT_plus_sigma2I_2d[1][1]
det = a * d - b * c
BBT_plus_sigma2I_inv_2d = [[d / det, -b / det], [-c / det, a / det]]

# Compute B^T (BB^T + σ^2 I)^(-1) (M x D)
BT_inv_2d = [[0.0 for _ in range(D_2d)] for _ in range(M_2d)]
for m in range(M_2d):
    for d2 in range(D_2d):
        for d1 in range(D_2d):
            BT_inv_2d[m][d2] += B_2d[d1][m] * BBT_plus_sigma2I_inv_2d[d1][d2]

# Compute m = B^T (BB^T + σ^2 I)^(-1) (x - μ)
m_2d = [0.0] * M_2d
for m in range(M_2d):
    for d in range(D_2d):
        m_2d[m] += BT_inv_2d[m][d] * x_minus_mu_2d[d]
print("\nPosterior Mean m:", m_2d)

# Compute C = I - B^T (BB^T + σ^2 I)^(-1) B
# First, compute (BB^T + σ^2 I)^(-1) B (D x M)
inv_B_2d = [[0.0 for _ in range(M_2d)] for _ in range(D_2d)]
for d1 in range(D_2d):
    for m in range(M_2d):
        for d2 in range(D_2d):
            inv_B_2d[d1][m] += BBT_plus_sigma2I_inv_2d[d1][d2] * B_2d[d2][m]

# Compute B^T (BB^T + σ^2 I)^(-1) B (M x M)
BT_inv_B_2d = [[0.0 for _ in range(M_2d)] for _ in range(M_2d)]
for m1 in range(M_2d):
    for m2 in range(M_2d):
        for d in range(D_2d):
            BT_inv_B_2d[m1][m2] += B_2d[d][m1] * inv_B_2d[d][m2]

# Compute C = I - B^T (BB^T + σ^2 I)^(-1) B
C_2d = [[1.0 if m1 == m2 else 0.0 for m2 in range(M_2d)] for m1 in range(M_2d)]
for m1 in range(M_2d):
    for m2 in range(M_2d):
        C_2d[m1][m2] -= BT_inv_B_2d[m1][m2]
print("Posterior Covariance C:")
for row in C_2d:
    print([f"{val:.3f}" for val in row])

# Compute determinant of C (for M=1, it's just C[0][0])
det_C_2d = C_2d[0][0]
print(f"Determinant of C: {det_C_2d:.3f}")

# Step 2: Sample from Posterior z* ~ N(m, C)
def sample_gaussian(mean, variance):
    u1, u2 = random.random(), random.random()
    z = math.sqrt(-2.0 * math.log(u1)) * math.cos(2.0 * math.pi * u2)
    return mean + math.sqrt(variance) * z

z_samples_2d = []
for _ in range(5):  # Sample 5 z* values
    z_star = [sample_gaussian(m_2d[0], C_2d[0][0])]
    z_samples_2d.append(z_star)
print("\nSampled z* from Posterior p(z|x*):")
for i, z in enumerate(z_samples_2d):
    print(f"z*_{i+1}: {z}")

# Step 3: Generate x̃* ~ N(B z* + μ, σ^2 I)
x_tilde_samples_2d = []
for z_star in z_samples_2d:
    # Compute B z*
    Bz = [0.0] * D_2d
    for d in range(D_2d):
        for m in range(M_2d):
            Bz[d] += B_2d[d][m] * z_star[m]
    # Add μ
    x_tilde = [Bz[d] + mu_2d[d] for d in range(D_2d)]
    # Add noise ϵ ~ N(0, σ^2 I)
    for d in range(D_2d):
        x_tilde[d] += sample_gaussian(0, sigma2_2d)
    x_tilde_samples_2d.append(x_tilde)
print("\nGenerated x̃* from p(x|z*, B, μ, σ^2):")
for i, x in enumerate(x_tilde_samples_2d):
    print(f"x̃*_{i+1}: {x}")

# --- Part 2: Posterior Distribution on Simulated MNIST-like Dataset ---
# Dataset (D = 4, N = 5, M = 2)
data_mnist = [
    [1.0, 2.0, 1.5, 2.5],
    [1.1, 2.1, 1.6, 2.4],
    [0.9, 1.9, 1.4, 2.6],
    [1.2, 2.2, 1.7, 2.3],
    [1.0, 2.0, 1.5, 2.5]
]
N_mnist = len(data_mnist)
D_mnist = len(data_mnist[0])
M_mnist = 2

# New observation x* (average of the dataset for simplicity)
x_star_mnist = [sum(col) / N_mnist for col in zip(*data_mnist)]  # Mean of the data

# Model parameters from prior implementation
mu_mnist = [1.04, 2.04, 1.54, 2.46]
B_mnist = [
    [0.5, 0.7071067811865475],
    [0.5, -0.7071067811865475],
    [0.5, 0.0],
    [-0.5, 0.0]
]
sigma2_mnist = -1.3877787807814457e-17  # Very small, set to small positive for stability
sigma2_mnist = max(sigma2_mnist, 1e-10)

print("\n=== Part 2: Posterior Distribution on Simulated MNIST-like Dataset ===")
print("New Observation x*:", x_star_mnist)

# Compute x - μ
x_minus_mu_mnist = [x_star_mnist[d] - mu_mnist[d] for d in range(D_mnist)]

# Compute BB^T (D x D)
BBT_mnist = [[0.0 for _ in range(D_mnist)] for _ in range(D_mnist)]
for d1 in range(D_mnist):
    for d2 in range(D_mnist):
        for m in range(M_mnist):
            BBT_mnist[d1][d2] += B_mnist[d1][m] * B_mnist[d2][m]

# Compute BB^T + σ^2 I
BBT_plus_sigma2I_mnist = [[BBT_mnist[d1][d2] + (sigma2_mnist if d1 == d2 else 0.0) for d2 in range(D_mnist)] for d1 in range(D_mnist)]

# Compute (BB^T + σ^2 I)^(-1) using a simple method (adjugate/determinant for 4x4)
# For simplicity, approximate by assuming small σ^2 and use numerical inverse (or manual computation)
# We'll compute it manually for 4x4 (or use a library in practice, but here we simplify)
def matrix_multiply(A, B, A_rows, A_cols, B_cols):
    result = [[0.0 for _ in range(B_cols)] for _ in range(A_rows)]
    for i in range(A_rows):
        for j in range(B_cols):
            for k in range(A_cols):
                result[i][j] += A[i][k] * B[k][j]
    return result

# Approximate inverse by assuming BB^T + σ^2 I is nearly BB^T (since σ^2 is tiny)
# In practice, use a proper numerical method; here we simplify for core Python
BBT_plus_sigma2I_inv_mnist = [[0.0 for _ in range(D_mnist)] for _ in range(D_mnist)]
# Since σ^2 is very small, approximate inverse manually (this is a simplification)
# For a proper implementation, we'd need a full 4x4 inverse, but let's compute key terms
# First, compute BB^T inverse approximately
BBT_det = 1.0  # Simplified (in practice, compute full determinant)
for d in range(D_mnist):
    BBT_plus_sigma2I_inv_mnist[d][d] = 1.0 / (BBT_mnist[d][d] + sigma2_mnist)

# Compute B^T (BB^T + σ^2 I)^(-1) (M x D)
BT_inv_mnist = [[0.0 for _ in range(D_mnist)] for _ in range(M_mnist)]
for m in range(M_mnist):
    for d2 in range(D_mnist):
        for d1 in range(D_mnist):
            BT_inv_mnist[m][d2] += B_mnist[d1][m] * BBT_plus_sigma2I_inv_mnist[d1][d2]

# Compute m = B^T (BB^T + σ^2 I)^(-1) (x - μ)
m_mnist = [0.0] * M_mnist
for m in range(M_mnist):
    for d in range(D_mnist):
        m_mnist[m] += BT_inv_mnist[m][d] * x_minus_mu_mnist[d]
print("\nPosterior Mean m:", m_mnist)

# Compute C = I - B^T (BB^T + σ^2 I)^(-1) B
inv_B_mnist = [[0.0 for _ in range(M_mnist)] for _ in range(D_mnist)]
for d1 in range(D_mnist):
    for m in range(M_mnist):
        for d2 in range(D_mnist):
            inv_B_mnist[d1][m] += BBT_plus_sigma2I_inv_mnist[d1][d2] * B_mnist[d2][m]

BT_inv_B_mnist = [[0.0 for _ in range(M_mnist)] for _ in range(M_mnist)]
for m1 in range(M_mnist):
    for m2 in range(M_mnist):
        for d in range(D_mnist):
            BT_inv_B_mnist[m1][m2] += B_mnist[d][m1] * inv_B_mnist[d][m2]

C_mnist = [[1.0 if m1 == m2 else 0.0 for m2 in range(M_mnist)] for m1 in range(M_mnist)]
for m1 in range(M_mnist):
    for m2 in range(M_mnist):
        C_mnist[m1][m2] -= BT_inv_B_mnist[m1][m2]
print("Posterior Covariance C:")
for row in C_mnist:
    print([f"{val:.3f}" for val in row])

# Compute determinant of C (for 2x2)
det_C_mnist = C_mnist[0][0] * C_mnist[1][1] - C_mnist[0][1] * C_mnist[1][0]
print(f"Determinant of C: {det_C_mnist:.3f}")

# Sample from Posterior z* ~ N(m, C) (simplified for 2x2)
# For simplicity, assume C is diagonal (due to numerical issues in this approximation)
z_samples_mnist = []
for _ in range(5):
    z_star = [
        sample_gaussian(m_mnist[0], max(C_mnist[0][0], 1e-10)),
        sample_gaussian(m_mnist[1], max(C_mnist[1][1], 1e-10))
    ]
    z_samples_mnist.append(z_star)
print("\nSampled z* from Posterior p(z|x*):")
for i, z in enumerate(z_samples_mnist):
    print(f"z*_{i+1}: {z}")

# Generate x̃* ~ N(B z* + μ, σ^2 I)
x_tilde_samples_mnist = []
for z_star in z_samples_mnist:
    Bz = [0.0] * D_mnist
    for d in range(D_mnist):
        for m in range(M_mnist):
            Bz[d] += B_mnist[d][m] * z_star[m]
    x_tilde = [Bz[d] + mu_mnist[d] for d in range(D_mnist)]
    for d in range(D_mnist):
        x_tilde[d] += sample_gaussian(0, sigma2_mnist)
    x_tilde_samples_mnist.append(x_tilde)
print("\nGenerated x̃* from p(x|z*, B, μ, σ^2):")
for i, x in enumerate(x_tilde_samples_mnist):
    print(f"x̃*_{i+1}: {x}")

# --- Plotting for Part 1 (2D Dataset) ---
def plot_posterior_2d(data, x_star, z_samples, x_tilde_samples):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    # Plot 1: Observed space
    data_x = [x[0] for x in data]
    data_y = [x[1] for x in data]
    x_tilde_x = [x[0] for x in x_tilde_samples]
    x_tilde_y = [x[1] for x in x_tilde_samples]
    ax1.scatter(data_x, data_y, color='blue', label='Original Data')
    ax1.scatter([x_star[0]], [x_star[1]], color='red', label='x*', marker='x')
    ax1.scatter(x_tilde_x, x_tilde_y, color='orange', label='Generated x̃*', marker='o', alpha=0.5)
    ax1.set_title('2D Data: Original, x*, and Generated')
    ax1.set_xlabel('x1')
    ax1.set_ylabel('x2')
    ax1.legend()
    ax1.grid(True)
    ax1.axis('equal')

    # Plot 2: Latent space (1D)
    z_values = [z[0] for z in z_samples]
    ax2.scatter(z_values, [0] * len(z_values), color='orange', label='Sampled z*', marker='o', alpha=0.5)
    ax2.scatter([m_2d[0]], [0], color='red', label='Posterior Mean m', marker='x')
    ax2.set_title('1D Latent Space: Posterior Samples')
    ax2.set_xlabel('z')
    ax2.set_yticks([])
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

# --- Plotting for Part 2 (MNIST-like Dataset) ---
def plot_posterior_mnist(z_samples, m):
    fig, ax = plt.subplots(figsize=(6, 6))
    z_x = [z[0] for z in z_samples]
    z_y = [z[1] for z in z_samples]
    ax.scatter(z_x, z_y, color='orange', label='Sampled z*', marker='o', alpha=0.5)
    ax.scatter([m[0]], [m[1]], color='red', label='Posterior Mean m', marker='x')
    ax.set_title('2D Latent Space: Posterior Samples for Simulated MNIST')
    ax.set_xlabel('z1')
    ax.set_ylabel('z2')
    ax.legend()
    ax.grid(True)
    ax.axis('equal')
    plt.show()

# --- Main Execution ---
if __name__ == "__main__":
    print("PPCA Posterior Distribution Implementation")
    print("=" * 60)

    # Part 1
    print("\nPlotting Posterior Results for 2D Dataset")
    plot_posterior_2d(data_2d, x_star_2d, z_samples_2d, x_tilde_samples_2d)

    # Part 2
    print("\nPlotting Posterior Results for Simulated MNIST Dataset")
    plot_posterior_mnist(z_samples_mnist, m_mnist)

    print("\n" + "=" * 60)
    print("Summary of Results:")
    print("• Computed posterior p(z|x*) for a new observation in both datasets.")
    print("• Sampled from the posterior and generated corresponding data points.")
    print("• Visualized the posterior samples in the latent and observed spaces.")
```
