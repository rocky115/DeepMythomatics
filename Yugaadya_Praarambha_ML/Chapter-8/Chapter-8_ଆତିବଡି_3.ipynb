###  MAP Estimation as Regularization

Instead of placing a prior distribution on the parameters $ \theta $, it is also possible to mitigate the effect of overfitting by penalizing the amplitude of the parameter by means of regularization. In regularized least squares, we consider the loss function

$$
\|\mathbf{y} - \Phi\theta\|^2 + \lambda \|\theta\|_2^2, \tag{9.32}
$$

which we minimize with respect to $ \theta $ (see Section 8.2.3). Here, the first term is a data-fit term (also called misfit term), which is proportional to the negative log-likelihood; see (9.10b). The second term is called the regularizer, and the regularization parameter $ \lambda \geq 0 $ controls the “strictness” of the regularization.

> **Remark**: Instead of the Euclidean norm $ \|\cdot\|_2 $, we can choose any $ p $-norm $ \|\cdot\|_p $ in (9.32). In practice, smaller values for $ p $ lead to sparser solutions. Here, “sparse” means that many parameter values $ \theta_d = 0 $, which is also useful for variable selection. For $ p = 1 $, the regularizer is called **LASSO** (least absolute shrinkage and selection operator) and was proposed by Tibshirani (1996). ♢

The regularizer $ \lambda \|\theta\|_2^2 $ in (9.32) can be interpreted as a negative log-Gaussian prior, which we use in MAP estimation; see (9.26). More specifically, with a Gaussian prior

$$
p(\theta) = \mathcal{N}(0, b^2 \mathbf{I}),
$$

we obtain the negative log-Gaussian prior

$$
-\log p(\theta) = \frac{1}{2b^2} \|\theta\|_2^2 + \text{const} \tag{9.33}
$$

so that for $ \lambda = \frac{1}{2b^2} $, the regularization term and the negative log-Gaussian prior are identical.

Given that the regularized least-squares loss function in (9.32) consists of terms that are closely related to the negative log-likelihood plus a negative log-prior, it is not surprising that, when we minimize this loss, we obtain a solution that closely resembles the MAP estimate in (9.31). More specifically, minimizing the regularized least-squares loss function yields

$$
\theta_{\text{RLS}} = (\Phi^\top \Phi + \lambda \mathbf{I})^{-1} \Phi^\top \mathbf{y}, \tag{9.34}
$$

which is identical to the MAP estimate in (9.31) for

$$
\lambda = \frac{\sigma^2}{b^2},
$$

where $ \sigma^2 $ is the noise variance and $ b^2 $ the variance of the (isotropic) Gaussian prior $ p(\theta) = \mathcal{N}(0, b^2 \mathbf{I}) $.

> **Remark**: A point estimate is a single specific parameter value, unlike a distribution over plausible parameter settings. ♢

So far, we have covered parameter estimation using maximum likelihood and MAP estimation where we found point estimates $ \theta^* $ that optimize an objective function (likelihood or posterior). We saw that both maximum likelihood and MAP estimation can lead to overfitting. In the next section, we will discuss Bayesian linear regression, where we use Bayesian inference (Section 8.4) to find a posterior distribution over the unknown parameters, which we subsequently use to make predictions. More specifically, for predictions we will average over all plausible sets of parameters instead of focusing on a point estimate.

###  Bayesian Linear Regression

Previously, we looked at linear regression models where we estimated the model parameters $ \theta $, e.g., by means of maximum likelihood or MAP estimation. We discovered that MLE can lead to severe overfitting, in particular, in the small-data regime. MAP addresses this issue by placing a prior on the parameters that plays the role of a regularizer. **Bayesian linear regression** pushes the idea of the parameter prior a step further and does not even attempt to compute a point estimate of the parameters, but instead the full posterior distribution over the parameters is taken into account when making predictions. This means we do not fit any parameters, but we compute a mean over all plausible parameters settings (according to the posterior).

```python
import math
import random
from typing import List, Tuple, Callable

class LinearRegressionWithFeatures:
    def __init__(self, sigma2: float = None, feature_transform: Callable[[List[float]], List[float]] = None, prior_b2: float = None):
        """
        Linear Regression model: y = phi(x)^T theta + epsilon, epsilon ~ N(0, sigma^2).
        sigma2 can be provided or estimated.
        prior_b2 is the variance of the Gaussian prior N(0, b^2 I) for MAP estimation.
        """
        self.sigma2 = sigma2
        self.feature_transform = feature_transform if feature_transform else lambda x: x
        self.prior_b2 = prior_b2
        self.theta = None
        self.fitted = False
        self.estimated_sigma2 = None
        self.fit_method = None

    def _compute_features(self, X: List[List[float]]) -> List[List[float]]:
        return [self.feature_transform(x) for x in X]

    def fit_mle(self, X: List[List[float]], y: List[float]):
        """
        Fit using Maximum Likelihood Estimation: theta_ML = (Phi^T Phi)^(-1) Phi^T y.
        """
        self.fit_method = "MLE"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True
        if self.sigma2 is None:
            self.estimate_noise_variance(X, y)

    def fit_map(self, X: List[List[float]], y: List[float]):
        """
        Fit using Maximum A Posteriori Estimation: theta_MAP = (Phi^T Phi + (sigma^2/b^2) I)^(-1) Phi^T y.
        """
        if self.prior_b2 is None:
            raise ValueError("Prior variance b^2 must be provided for MAP estimation")
        if self.sigma2 is None:
            self.fit_mle(X, y)
            self.sigma2 = self.estimated_sigma2
            self.fitted = False

        self.fit_method = "MAP"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        lambda_reg = self.sigma2 / self.prior_b2
        for i in range(n_features):
            PhiT_Phi[i][i] += lambda_reg

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True

    def fit_rls(self, X: List[List[float]], y: List[float], lambda_reg: float):
        """
        Fit using Regularized Least Squares: theta_RLS = (Phi^T Phi + lambda I)^(-1) Phi^T y.
        """
        self.fit_method = "RLS"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        for i in range(n_features):
            PhiT_Phi[i][i] += lambda_reg

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True

    def predict(self, X: List[List[float]]) -> List[float]:
        if not self.fitted:
            raise ValueError("Model must be fitted before prediction")
        Phi = self._compute_features(X)
        predictions = []
        for phi in Phi:
            pred = sum(t * p for t, p in zip(self.theta, phi))
            predictions.append(pred)
        return predictions

    def estimate_noise_variance(self, X: List[List[float]], y: List[float]) -> float:
        if not self.fitted:
            self.fit_mle(X, y)
        predictions = self.predict(X)
        n_samples = len(y)
        squared_errors = sum((yi - pred)**2 for yi, pred in zip(y, predictions))
        self.estimated_sigma2 = squared_errors / n_samples
        return self.estimated_sigma2

    def compute_rmse(self, X: List[List[float]], y: List[float]) -> float:
        predictions = self.predict(X)
        n_samples = len(y)
        mse = sum((yi - pred)**2 for yi, pred in zip(y, predictions)) / n_samples
        return math.sqrt(mse)

    def get_theta(self) -> List[float]:
        if not self.fitted:
            raise ValueError("Model must be fitted to access theta")
        return self.theta

    def _matrix_inverse(self, matrix: List[List[float]]) -> List[List[float]]:
        n = len(matrix)
        augmented = []
        for i in range(n):
            row = matrix[i][:] + [0.0] * n
            row[n + i] = 1.0
            augmented.append(row)
        for i in range(n):
            max_row = i
            for k in range(i + 1, n):
                if abs(augmented[k][i]) > abs(augmented[max_row][i]):
                    max_row = k
            augmented[i], augmented[max_row] = augmented[max_row], augmented[i]
            pivot = augmented[i][i]
            if abs(pivot) < 1e-10:
                raise ValueError("Matrix is singular")
            for j in range(2 * n):
                augmented[i][j] /= pivot
            for k in range(n):
                if k != i:
                    factor = augmented[k][i]
                    for j in range(2 * n):
                        augmented[k][j] -= factor * augmented[i][j]
        inverse = []
        for i in range(n):
            inverse.append(augmented[i][n:])
        return inverse

def polynomial_features(degree: int) -> Callable[[List[float]], List[float]]:
    def transform(x: List[float]) -> List[float]:
        x_val = x[0] if x else 0.0
        return [x_val ** k for k in range(degree + 1)]
    return transform

def generate_data(n_samples: int = 10, sigma2: float = 0.04) -> Tuple[List[List[float]], List[float]]:
    random.seed(42)
    X = []
    y = []
    for _ in range(n_samples):
        x_val = random.uniform(-5, 5)
        y_val = -math.sin(x_val / 5) + math.cos(x_val)
        noise = random.gauss(0, math.sqrt(sigma2))
        y_val += noise
        X.append([x_val])
        y.append(y_val)
    return X, y

def generate_test_data(n_samples: int = 200, sigma2: float = 0.04) -> Tuple[List[List[float]], List[float]]:
    random.seed(43)
    X = [[x] for x in [((i / (n_samples - 1)) * 10 - 5) for i in range(n_samples)]]
    y = []
    for x in X:
        x_val = x[0]
        y_val = -math.sin(x_val / 5) + math.cos(x_val)
        noise = random.gauss(0, math.sqrt(sigma2))
        y_val += noise
        y.append(y_val)
    return X, y

def main():
    print("="*60)
    print("LINEAR REGRESSION: REGULARIZED LEAST SQUARES (SECTION 9.2.4)")
    print("="*60)

    # Generate training and test data
    sigma2 = 0.04
    X_train, y_train = generate_data(n_samples=10, sigma2=sigma2)
    X_test, y_test = generate_test_data(n_samples=200, sigma2=sigma2)
    print(f"Training data: {len(X_train)} points")
    print(f"Test data: {len(X_test)} points")

    # Use degree 6 as in previous examples
    degree = 6
    print(f"\nFitting Polynomial of Degree {degree}...")
    feature_transform = polynomial_features(degree=degree)

    # MLE Model
    mle_model = LinearRegressionWithFeatures(sigma2=sigma2, feature_transform=feature_transform)
    try:
        mle_model.fit_mle(X_train, y_train)
        mle_rmse_train = mle_model.compute_rmse(X_train, y_train)
        mle_rmse_test = mle_model.compute_rmse(X_test, y_test)
        mle_theta = mle_model.get_theta()
        print(f"   MLE - Train RMSE: {mle_rmse_train:.4f}, Test RMSE: {mle_rmse_test:.4f}")
        print(f"   MLE Theta: {[round(t, 4) for t in mle_theta]}")
    except ValueError as e:
        print(f"   MLE - Error: {str(e)}")
        mle_model = None

    # MAP Model
    prior_b2 = 1.0  # p(theta) = N(0, I)
    map_model = LinearRegressionWithFeatures(sigma2=sigma2, feature_transform=feature_transform, prior_b2=prior_b2)
    try:
        map_model.fit_map(X_train, y_train)
        map_rmse_train = map_model.compute_rmse(X_train, y_train)
        map_rmse_test = map_model.compute_rmse(X_test, y_test)
        map_theta = map_model.get_theta()
        print(f"   MAP - Train RMSE: {map_rmse_train:.4f}, Test RMSE: {map_rmse_test:.4f}")
        print(f"   MAP Theta: {[round(t, 4) for t in map_theta]}")
    except ValueError as e:
        print(f"   MAP - Error: {str(e)}")
        map_model = None

    # RLS Model
    lambda_reg = sigma2 / prior_b2  # lambda = sigma^2 / b^2
    print(f"   Lambda (sigma^2/b^2): {lambda_reg:.4f}")
    rls_model = LinearRegressionWithFeatures(feature_transform=feature_transform)
    try:
        rls_model.fit_rls(X_train, y_train, lambda_reg)
        rls_rmse_train = rls_model.compute_rmse(X_train, y_train)
        rls_rmse_test = rls_model.compute_rmse(X_test, y_test)
        rls_theta = rls_model.get_theta()
        print(f"   RLS - Train RMSE: {rls_rmse_train:.4f}, Test RMSE: {rls_rmse_test:.4f}")
        print(f"   RLS Theta: {[round(t, 4) for t in rls_theta]}")
    except ValueError as e:
        print(f"   RLS - Error: {str(e)}")
        rls_model = None

    # Verify equivalence between MAP and RLS
    if map_model and rls_model:
        theta_diff = [abs(mt - rt) for mt, rt in zip(map_theta, rls_theta)]
        max_diff = max(theta_diff)
        print(f"\n   Max difference between MAP and RLS theta: {max_diff:.6f}")
        if max_diff < 1e-6:
            print("   MAP and RLS are equivalent (within numerical precision).")

if __name__ == "__main__":
    main()
```

```python
import math
import random
from typing import List, Tuple, Callable

import matplotlib.pyplot as plt

class LinearRegressionWithFeatures:
    def __init__(self, sigma2: float = None, feature_transform: Callable[[List[float]], List[float]] = None, prior_b2: float = None):
        self.sigma2 = sigma2
        self.feature_transform = feature_transform if feature_transform else lambda x: x
        self.prior_b2 = prior_b2
        self.theta = None
        self.fitted = False
        self.estimated_sigma2 = None
        self.fit_method = None

    def _compute_features(self, X: List[List[float]]) -> List[List[float]]:
        return [self.feature_transform(x) for x in X]

    def fit_mle(self, X: List[List[float]], y: List[float]):
        self.fit_method = "MLE"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True
        if self.sigma2 is None:
            self.estimate_noise_variance(X, y)

    def fit_map(self, X: List[List[float]], y: List[float]):
        if self.prior_b2 is None:
            raise ValueError("Prior variance b^2 must be provided for MAP estimation")
        if self.sigma2 is None:
            self.fit_mle(X, y)
            self.sigma2 = self.estimated_sigma2
            self.fitted = False

        self.fit_method = "MAP"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        lambda_reg = self.sigma2 / self.prior_b2
        for i in range(n_features):
            PhiT_Phi[i][i] += lambda_reg

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True

    def fit_rls(self, X: List[List[float]], y: List[float], lambda_reg: float):
        self.fit_method = "RLS"
        Phi = self._compute_features(X)
        n_samples = len(Phi)
        n_features = len(Phi[0])

        PhiT_Phi = [[0.0] * n_features for _ in range(n_features)]
        for i in range(n_features):
            for j in range(n_features):
                PhiT_Phi[i][j] = sum(Phi[k][i] * Phi[k][j] for k in range(n_samples))

        for i in range(n_features):
            PhiT_Phi[i][i] += lambda_reg

        PhiT_y = [0.0] * n_features
        for i in range(n_features):
            PhiT_y[i] = sum(Phi[k][i] * y[k] for k in range(n_samples))

        PhiT_Phi_inv = self._matrix_inverse(PhiT_Phi)
        self.theta = [0.0] * n_features
        for i in range(n_features):
            self.theta[i] = sum(PhiT_Phi_inv[i][j] * PhiT_y[j] for j in range(n_features))

        self.fitted = True

    def predict(self, X: List[List[float]]) -> List[float]:
        if not self.fitted:
            raise ValueError("Model must be fitted before prediction")
        Phi = self._compute_features(X)
        predictions = []
        for phi in Phi:
            pred = sum(t * p for t, p in zip(self.theta, phi))
            predictions.append(pred)
        return predictions

    def estimate_noise_variance(self, X: List[List[float]], y: List[float]) -> float:
        if not self.fitted:
            self.fit_mle(X, y)
        predictions = self.predict(X)
        n_samples = len(y)
        squared_errors = sum((yi - pred)**2 for yi, pred in zip(y, predictions))
        self.estimated_sigma2 = squared_errors / n_samples
        return self.estimated_sigma2

    def compute_rmse(self, X: List[List[float]], y: List[float]) -> float:
        predictions = self.predict(X)
        n_samples = len(y)
        mse = sum((yi - pred)**2 for yi, pred in zip(y, predictions)) / n_samples
        return math.sqrt(mse)

    def get_theta(self) -> List[float]:
        if not self.fitted:
            raise ValueError("Model must be fitted to access theta")
        return self.theta

    def _matrix_inverse(self, matrix: List[List[float]]) -> List[List[float]]:
        n = len(matrix)
        augmented = []
        for i in range(n):
            row = matrix[i][:] + [0.0] * n
            row[n + i] = 1.0
            augmented.append(row)
        for i in range(n):
            max_row = i
            for k in range(i + 1, n):
                if abs(augmented[k][i]) > abs(augmented[max_row][i]):
                    max_row = k
            augmented[i], augmented[max_row] = augmented[max_row], augmented[i]
            pivot = augmented[i][i]
            if abs(pivot) < 1e-10:
                raise ValueError("Matrix is singular")
            for j in range(2 * n):
                augmented[i][j] /= pivot
            for k in range(n):
                if k != i:
                    factor = augmented[k][i]
                    for j in range(2 * n):
                        augmented[k][j] -= factor * augmented[i][j]
        inverse = []
        for i in range(n):
            inverse.append(augmented[i][n:])
        return inverse

def polynomial_features(degree: int) -> Callable[[List[float]], List[float]]:
    def transform(x: List[float]) -> List[float]:
        x_val = x[0] if x else 0.0
        return [x_val ** k for k in range(degree + 1)]
    return transform

def generate_data(n_samples: int = 10, sigma2: float = 0.04) -> Tuple[List[List[float]], List[float]]:
    random.seed(42)
    X = []
    y = []
    for _ in range(n_samples):
        x_val = random.uniform(-5, 5)
        y_val = -math.sin(x_val / 5) + math.cos(x_val)
        noise = random.gauss(0, math.sqrt(sigma2))
        y_val += noise
        X.append([x_val])
        y.append(y_val)
    return X, y

def generate_test_data(n_samples: int = 200, sigma2: float = 0.04) -> Tuple[List[List[float]], List[float]]:
    random.seed(43)
    X = [[x] for x in [((i / (n_samples - 1)) * 10 - 5) for i in range(n_samples)]]
    y = []
    for x in X:
        x_val = x[0]
        y_val = -math.sin(x_val / 5) + math.cos(x_val)
        noise = random.gauss(0, math.sqrt(sigma2))
        y_val += noise
        y.append(y_val)
    return X, y

def plot_fits(X_train, y_train, X_test, mle_predictions, map_predictions, rls_predictions, degree):
    x_test_values = [x[0] for x in X_test]
    x_train_values = [x[0] for x in X_train]

    plt.figure(figsize=(8, 6))
    plt.scatter(x_train_values, y_train, color='blue', label='Training Data', s=50)
    plt.plot(x_test_values, mle_predictions, color='red', label='MLE Fit', linewidth=2)
    plt.plot(x_test_values, map_predictions, color='green', label='MAP Fit', linewidth=2)
    plt.plot(x_test_values, rls_predictions, color='purple', label='RLS Fit', linewidth=2, linestyle='--')
    plt.xlim(-5, 5)
    plt.ylim(-4, 4)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Polynomial Regression (Degree {degree}): MLE vs MAP vs RLS')
    plt.grid(True)
    plt.legend()
    plt.show()

def main():
    print("="*60)
    print("LINEAR REGRESSION: REGULARIZED LEAST SQUARES PLOT (SECTION 9.2.4)")
    print("="*60)

    sigma2 = 0.04
    X_train, y_train = generate_data(n_samples=10, sigma2=sigma2)
    X_test, y_test = generate_test_data(n_samples=200, sigma2=sigma2)
    print(f"Training data: {len(X_train)} points")
    print(f"Test data: {len(X_test)} points")

    degree = 6
    print(f"\nFitting Polynomial of Degree {degree}...")
    feature_transform = polynomial_features(degree=degree)

    mle_model = LinearRegressionWithFeatures(sigma2=sigma2, feature_transform=feature_transform)
    try:
        mle_model.fit_mle(X_train, y_train)
        mle_predictions = mle_model.predict(X_test)
        mle_rmse_train = mle_model.compute_rmse(X_train, y_train)
        mle_rmse_test = mle_model.compute_rmse(X_test, y_test)
        print(f"   MLE - Train RMSE: {mle_rmse_train:.4f}, Test RMSE: {mle_rmse_test:.4f}")
    except ValueError as e:
        print(f"   MLE - Error: {str(e)}")
        mle_predictions = [0] * len(X_test)

    prior_b2 = 1.0
    map_model = LinearRegressionWithFeatures(sigma2=sigma2, feature_transform=feature_transform, prior_b2=prior_b2)
    try:
        map_model.fit_map(X_train, y_train)
        map_predictions = map_model.predict(X_test)
        map_rmse_train = map_model.compute_rmse(X_train, y_train)
        map_rmse_test = map_model.compute_rmse(X_test, y_test)
        print(f"   MAP - Train RMSE: {map_rmse_train:.4f}, Test RMSE: {map_rmse_test:.4f}")
    except ValueError as e:
        print(f"   MAP - Error: {str(e)}")
        map_predictions = [0] * len(X_test)

    lambda_reg = sigma2 / prior_b2
    print(f"   Lambda (sigma^2/b^2): {lambda_reg:.4f}")
    rls_model = LinearRegressionWithFeatures(feature_transform=feature_transform)
    try:
        rls_model.fit_rls(X_train, y_train, lambda_reg)
        rls_predictions = rls_model.predict(X_test)
        rls_rmse_train = rls_model.compute_rmse(X_train, y_train)
        rls_rmse_test = rls_model.compute_rmse(X_test, y_test)
        print(f"   RLS - Train RMSE: {rls_rmse_train:.4f}, Test RMSE: {rls_rmse_test:.4f}")
    except ValueError as e:
        print(f"   RLS - Error: {str(e)}")
        rls_predictions = [0] * len(X_test)

    plot_fits(X_train, y_train, X_test, mle_predictions, map_predictions, rls_predictions, degree)

if __name__ == "__main__":
    main()
```

### Model

In Bayesian linear regression, we consider the model

$$
\text{prior: } p(\theta) = \mathcal{N}(\mathbf{m}_0, \mathbf{S}_0), \tag{9.35}
$$

$$
\text{likelihood: } p(y \mid \mathbf{x}, \theta) = \mathcal{N}(y \mid \phi^\top(\mathbf{x})\theta, \sigma^2),
$$

where we now explicitly place a Gaussian prior $ p(\theta) = \mathcal{N}(\mathbf{m}_0, \mathbf{S}_0) $ on $ \theta $, which turns the parameter vector into a random variable. This allows us to write down the corresponding graphical model in Fig.8, where we made the parameters of the Gaussian prior on $ \theta $ explicit. The full probabilistic model, i.e., the joint distribution of observed and unobserved random variables, $ y $ and $ \theta $, respectively, is

$$
p(y, \theta \mid \mathbf{x}) = p(y \mid \mathbf{x}, \theta) p(\theta). \tag{9.36}
$$

#### Prior Predictions

In practice, we are usually not so much interested in the parameter values $ \theta $ themselves. Instead, our focus often lies in the predictions we make with those parameter values. In a Bayesian setting, we take the parameter distribution and average over all plausible parameter settings when we make predictions. More specifically, to make predictions at an input $ \mathbf{x}^* $, we integrate out $ \theta $ and obtain

$$
p(y^* \mid \mathbf{x}^*) = \int p(y^* \mid \mathbf{x}^*, \theta) p(\theta) \, d\theta = \mathbb{E}_\theta[p(y^* \mid \mathbf{x}^*, \theta)], \tag{9.37}
$$

which we can interpret as the average prediction of $ y^* \mid \mathbf{x}^*, \theta $ for all plausible parameters $ \theta $ according to the prior distribution $ p(\theta) $. Note that predictions using the prior distribution only require us to specify the input $ \mathbf{x}^* $, but no training data.

In our model (9.35), we chose a conjugate (Gaussian) prior on $ \theta $ so that the predictive distribution is Gaussian as well (and can be computed in closed form): With the prior distribution $ p(\theta) = \mathcal{N}(\mathbf{m}_0, \mathbf{S}_0) $, we obtain the predictive distribution as

$$
p(y^* \mid \mathbf{x}^*) = \mathcal{N}(\phi^\top(\mathbf{x}^*)\mathbf{m}_0, \phi^\top(\mathbf{x}^*)\mathbf{S}_0 \phi(\mathbf{x}^*) + \sigma^2), \tag{9.38}
$$

where we exploited that (i) the prediction is Gaussian due to conjugacy (see Section 6.6) and the marginalization property of Gaussians (see Section 6.5), (ii) the Gaussian noise is independent so that

$$
\text{Var}[y^*] = \text{Var}_\theta[\phi^\top(\mathbf{x}^*)\theta] + \text{Var}_\epsilon[\epsilon], \tag{9.39}
$$

and (iii) $ y^* $ is a linear transformation of $ \theta $ so that we can apply the rules for computing the mean and covariance of the prediction analytically by using (6.50) and (6.51), respectively.

In (9.38), the term $ \phi^\top(\mathbf{x}^*)\mathbf{S}_0 \phi(\mathbf{x}^*) $ in the predictive variance explicitly accounts for the uncertainty associated with the parameters $ \theta $, whereas $ \sigma^2 $ is the uncertainty contribution due to the measurement noise.

If we are interested in predicting noise-free function values $ f(\mathbf{x}^*) = \phi^\top(\mathbf{x}^*)\theta $ instead of the noise-corrupted targets $ y^* $, we obtain

$$
p(f(\mathbf{x}^*)) = \mathcal{N}(\phi^\top(\mathbf{x}^*)\mathbf{m}_0, \phi^\top(\mathbf{x}^*)\mathbf{S}_0 \phi(\mathbf{x}^*)), \tag{9.40}
$$

which only differs from (9.38) in the omission of the noise variance $ \sigma^2 $ in the predictive variance.

> **Remark (Distribution over Functions)**: Since we can represent the distribution $ p(\theta) $ using a set of samples $ \theta^i $, and every sample $ \theta^i $ gives rise to a function $ f_i(\cdot) = (\theta^i)^\top \phi(\cdot) $, it follows that the parameter distribution $ p(\theta) $ induces a distribution $ p(f(\cdot)) $ over functions. Here we use the notation $ (\cdot) $ to explicitly denote a functional relationship. The parameter distribution $ p(\theta) $ induces a distribution over functions. ♢

### Example 7 (Prior over Functions)


**Fig.9**: Prior over functions.

(a) Distribution over functions represented by the mean function (black line) and the marginal uncertainties (shaded), representing the 67% and 95% confidence bounds, respectively;  
(b) Samples from the prior distribution over functions.

Let us consider a Bayesian linear regression problem with polynomials of degree 5. We choose a parameter prior

$$
p(\theta) = \mathcal{N}\left(0, \frac{1}{4} \mathbf{I}\right).
$$

Figure 9.9 visualizes the induced prior distribution over functions (shaded area: dark gray: 67% confidence bound; light gray: 95% confidence bound) induced by this parameter prior, including some function samples from this prior. A function sample is obtained by first sampling a parameter vector $ \theta^i \sim p(\theta) $ and then computing $ f_i(\cdot) = (\theta^i)^\top \phi(\cdot) $. We used 200 input locations $ \mathbf{x}^* \in [-5, 5] $ to which we apply the feature function $ \phi(\cdot) $. The uncertainty (represented by the shaded area) in Figure 9.9 is solely due to the parameter uncertainty because we considered the noise-free predictive distribution (9.40).

So far, we looked at computing predictions using the parameter prior $ p(\theta) $. However, when we have a parameter posterior (given some training data $ \mathbf{X}, \mathbf{Y} $), the same principles for prediction and inference hold as in (9.37) – we just need to replace the prior $ p(\theta) $ with the posterior $ p(\theta \mid \mathbf{X}, \mathbf{Y}) $. In the following, we will derive the posterior distribution in detail before using it to make predictions.

```python
import math
import random
from typing import List, Callable, Tuple  # Added Tuple to the imports

import matplotlib.pyplot as plt

class BayesianLinearRegression:
    def __init__(self, sigma2: float, m0: List[float], S0: List[List[float]], feature_transform: Callable[[List[float]], List[float]]):
        """
        Bayesian Linear Regression with prior p(theta) = N(m0, S0).
        sigma2: noise variance
        m0: prior mean for theta
        S0: prior covariance for theta
        feature_transform: function to transform inputs into features
        """
        self.sigma2 = sigma2
        self.m0 = m0
        self.S0 = S0
        self.feature_transform = feature_transform

    def _compute_features(self, X: List[List[float]]) -> List[List[float]]:
        return [self.feature_transform(x) for x in X]

    def _matrix_vector_multiply(self, matrix: List[List[float]], vector: List[float]) -> List[float]:
        result = [0.0] * len(matrix)
        for i in range(len(matrix)):
            result[i] = sum(matrix[i][j] * vector[j] for j in range(len(vector)))
        return result

    def _vector_matrix_vector_multiply(self, vector: List[float], matrix: List[List[float]]) -> float:
        temp = [0.0] * len(matrix)
        for i in range(len(matrix)):
            temp[i] = sum(matrix[i][j] * vector[j] for j in range(len(vector)))
        return sum(vector[i] * temp[i] for i in range(len(vector)))

    def prior_predictive(self, X: List[List[float]], noise_free: bool = False) -> Tuple[List[float], List[float]]:  # Fixed return type
        """
        Compute the prior predictive distribution p(y*|x*) or p(f(x*)).
        Returns (means, variances) for each input in X.
        If noise_free=True, compute p(f(x*)) without noise variance.
        """
        Phi = self._compute_features(X)
        means = []
        variances = []

        for phi in Phi:
            # Mean: phi^T m0
            mean = sum(phi[i] * self.m0[i] for i in range(len(self.m0)))
            means.append(mean)

            # Variance: phi^T S0 phi + sigma^2 (if not noise-free)
            var = self._vector_matrix_vector_multiply(phi, self.S0)
            if not noise_free:
                var += self.sigma2
            variances.append(var)

        return means, variances

    def sample_theta(self) -> List[float]:
        """
        Sample theta from the prior p(theta) = N(m0, S0) using the Cholesky decomposition method.
        Assumes S0 is diagonal for simplicity in core Python.
        """
        if not all(self.S0[i][j] == 0 for i in range(len(self.S0)) for j in range(len(self.S0[i])) if i != j):
            raise ValueError("This implementation assumes a diagonal S0 for simplicity")

        theta = []
        for i in range(len(self.m0)):
            # For a diagonal S0, sqrt(S0[i][i]) is the standard deviation
            std = math.sqrt(self.S0[i][i])
            theta_i = self.m0[i] + random.gauss(0, std)
            theta.append(theta_i)
        return theta

    def predict_with_theta(self, X: List[List[float]], theta: List[float]) -> List[float]:
        """
        Compute predictions f(x) = phi^T(x) theta for a given theta.
        """
        Phi = self._compute_features(X)
        predictions = []
        for phi in Phi:
            pred = sum(phi[i] * theta[i] for i in range(len(theta)))
            predictions.append(pred)
        return predictions

def polynomial_features(degree: int) -> Callable[[List[float]], List[float]]:
    def transform(x: List[float]) -> List[float]:
        x_val = x[0] if x else 0.0
        return [x_val ** k for k in range(degree + 1)]
    return transform

def generate_inputs(n_samples: int = 200) -> List[List[float]]:
    """
    Generate input points x in [-5, 5] for prediction.
    """
    return [[((i / (n_samples - 1)) * 10 - 5)] for i in range(n_samples)]

def plot_prior_predictions(X: List[List[float]], means: List[float], variances: List[float], sample_functions: List[List[float]]):
    x_values = [x[0] for x in X]

    # Figure 9.9(a): Mean and confidence bounds
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    # Compute standard deviations for confidence bounds
    stds = [math.sqrt(var) for var in variances]
    # 67% confidence interval (~1 std)
    y_upper_67 = [m + s for m, s in zip(means, stds)]
    y_lower_67 = [m - s for m, s in zip(means, stds)]
    # 95% confidence interval (~2 std)
    y_upper_95 = [m + 2*s for m, s in zip(means, stds)]
    y_lower_95 = [m - 2*s for m, s in zip(means, stds)]

    # Plot shaded areas for confidence bounds
    plt.fill_between(x_values, y_lower_95, y_upper_95, color='lightgray', alpha=0.5, label='95% Confidence')
    plt.fill_between(x_values, y_lower_67, y_upper_67, color='gray', alpha=0.5, label='67% Confidence')
    # Plot mean function
    plt.plot(x_values, means, color='black', label='Mean Function', linewidth=2)
    plt.xlim(-5, 5)
    plt.ylim(-4, 4)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Prior Distribution over Functions')
    plt.grid(True)
    plt.legend()

    # Figure 9.9(b): Sample functions
    plt.subplot(1, 2, 2)
    colors = ['blue', 'red', 'green', 'purple', 'orange']
    for i, sample_func in enumerate(sample_functions):
        plt.plot(x_values, sample_func, color=colors[i], label=f'Sample {i+1}', linewidth=1.5)
    plt.xlim(-5, 5)
    plt.ylim(-4, 4)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Samples from Prior Distribution')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

def main():
    print("="*60)
    print("BAYESIAN LINEAR REGRESSION: PRIOR PREDICTIONS (SECTION 9.3.2)")
    print("="*60)

    # Parameters from Example 9.7
    degree = 5  # Polynomial degree
    sigma2 = 0.04  # Noise variance (not used for noise-free predictions)
    prior_variance = 1/4  # S0 = (1/4) I
    n_features = degree + 1  # Number of parameters (theta_0 to theta_5)

    # Prior mean m0 = 0
    m0 = [0.0] * n_features
    # Prior covariance S0 = (1/4) I (diagonal for simplicity)
    S0 = [[0.0] * n_features for _ in range(n_features)]
    for i in range(n_features):
        S0[i][i] = prior_variance

    # Initialize model
    feature_transform = polynomial_features(degree=degree)
    model = BayesianLinearRegression(sigma2=sigma2, m0=m0, S0=S0, feature_transform=feature_transform)

    # Generate input points
    X = generate_inputs(n_samples=200)
    print(f"Input points for prediction: {len(X)}")

    # Compute prior predictive distribution (noise-free, as in Example 9.7)
    means, variances = model.prior_predictive(X, noise_free=True)
    print(f"Computed prior predictive distribution (noise-free)")

    # Sample 5 functions from the prior
    sample_functions = []
    random.seed(42)  # For reproducibility
    for i in range(5):
        theta_sample = model.sample_theta()
        sample_func = model.predict_with_theta(X, theta_sample)
        sample_functions.append(sample_func)
        print(f"Sampled function {i+1} from prior")

    # Plot results
    plot_prior_predictions(X, means, variances, sample_functions)

if __name__ == "__main__":
    main()
```
