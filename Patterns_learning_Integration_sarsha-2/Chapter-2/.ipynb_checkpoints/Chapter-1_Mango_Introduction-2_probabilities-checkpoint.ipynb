### 2 Probability Theory

A key concept in the field of pattern recognition is that of uncertainty. It arises both through noise on measurements, as well as through the finite size of data sets. Probability theory provides a consistent framework for the quantification and manipulation of uncertainty and forms one of the central foundations for pattern recognition. When combined with decision theory, discussed in Section 1.5, it allows us to make optimal predictions given all the information available to us, even though that information may be incomplete or ambiguous.

We will introduce the basic concepts of probability theory by considering a simple example. Imagine we have two boxes, one red and one blue. In the red box, we have 2 apples and 6 oranges, and in the blue box, we have 3 apples and 1 orange. This is illustrated in **Figure.9**.



Figure.9 We use a simple example of two coloured boxes each containing fruit (apples shown in green and or- anges shown in orange) to intro- duce the basic ideas of probability.

Now suppose we randomly pick one of the boxes, and from that box, we randomly select an item of fruit, and after observing the fruit, we replace it in the box from which it came. We could imagine repeating this process many times. Let us suppose that we pick the red box 40% of the time and the blue box 60% of the time, and that when we remove an item of fruit from a box, we are equally likely to select any of the pieces of fruit in the box.

In this example, the identity of the box that will be chosen is a random variable, denoted by \(B\). This random variable can take one of two possible values: \(r\) (corresponding to the red box) or $b$ (corresponding to the blue box). Similarly, the identity of the fruit is also a random variable, denoted by \(F\), which can take either \(a\) (for apple) or $o$ (for orange).

To begin, we define the probability of an event as the fraction of times that event occurs out of the total number of trials, in the limit that the total number of trials goes to infinity. Thus, the probability of selecting the red box is:

$$ p(B = r) = \frac{4}{10} $$

and the probability of selecting the blue box is:

$$ p(B = b) = \frac{6}{10} $$

Note that, by definition, probabilities must lie in the interval $[0, 1]$. Also, if the events are mutually exclusive and if they include all possible outcomes (for instance, the box must be either red or blue), then the probabilities for those events must sum to one.

We can now ask questions such as: "What is the overall probability that the selection procedure will pick an apple?" or "Given that we have chosen an orange, what is the probability that the box we chose was the blue one?". We can answer such questions using the two elementary rules of probability: the **sum rule** and the **product rule**.



Figure.10 We can derive the sum and product rules of probability by considering two random variables, X, which takes the values {$x_i$ } where i = 1, . . . , M , and Y , which takes the values {$y_j$ } where j = 1, . . . , L. In this illustration we have M = 5 and L = 3. If we consider a total number N of instances of these variables, then we denote the number of instances where $X = x_i$ and $Y = y_j$ by $n_ij$ , which is the number of points in the corresponding cell of the array. The number of points in column i, corresponding to $X = x_i$ , is denoted by $c_i$ , and the number of points in row j, corresponding to $Y = y_j$ , is denoted by $r_j$ .

In order to derive the rules of probability, consider the slightly more general example shown in **Figure.10** involving two random variables \(X\) and \(Y\). We assume \(X\) can take any of the values $x_i$, where $i = 1, \dots, M$, and $Y$ can take the values $y_j$, where $j = 1, \dots, L$.

Consider a total of $N$ trials in which we sample both variables $X$ and $Y$, and let the number of such trials in which $X = x_i$ and $Y = y_j$ be $n_{ij}$. Let the number of trials in which \(X\) takes the value $x_i$ (irrespective of $Y$'s value) be denoted by $c_i$, and the number of trials in which $Y$ takes the value $y_j$ be denoted by $r_j$.

The **joint probability** of $X = x_i$ and $Y = y_j$ is given by:

$$ p(X = x_i, Y = y_j) = \frac{n_{ij}}{N} \tag{1.5} $$

Here, we are considering the limit as $N \to \infty$. Similarly, the probability that $X$ takes the value $x_i$ irrespective of $Y$'s value is written as:

$$ p(X = x_i) = \frac{c_i}{N} \tag{1.6} $$

Since the number of instances in column \(i\) is just the sum of the number of instances in each cell of that column, we have:

$$ c_i = \sum_j n_{ij} $$




From equation (1.5) and (1.6), we have:

$$
p(X = x_i) = \sum_{j=1}^{L} p(X = x_i, Y = y_j) \tag{1.7}
$$

which is the **sum rule of probability**. Note that $p(X = x_i)$ is sometimes called the **marginal probability**, because it is obtained by marginalizing, or summing out, the other variables (in this case $Y$).

If we consider only those instances for which $X = x_i$, then the fraction of such instances for which $Y = y_j$ is written as $p(Y = y_j | X = x_i)$ and is called the **conditional probability** of $Y = y_j$ given $X = x_i$. It is given by:

$$
p(Y = y_j | X = x_i) = \frac{n_{ij}}{c_i} \tag{1.8}
$$

From equations (1.5), (1.6), and (1.8), we can derive the following relationship:

$$
p(X = x_i, Y = y_j) = \frac{n_{ij}}{N} = \frac{n_{ij}}{c_i} \cdot \frac{c_i}{N} = p(Y = y_j | X = x_i) p(X = x_i) \tag{1.9}
$$

which is the **product rule of probability**.

So far, we have carefully distinguished between a random variable, such as the box \(B\) in the fruit example, and the values that the random variable can take. For example, \(r\) if the box were the red one. Thus, the probability that \(B\) takes the value \(r\) is denoted \(p(B = r)\).

Although this avoids ambiguity, the notation can become cumbersome. Therefore, in many cases, we may write \(p(B)\) to denote a distribution over the random variable \(B\), or \(p(r)\) to denote the distribution evaluated for the particular value \(r\), provided the interpretation is clear from context.

With this more compact notation, we can write the two fundamental rules of probability theory in the following form:

**Sum rule:**

$$
p(X) = \sum_Y p(X, Y) \tag{1.10}
$$

**Product rule:**

$$
p(X, Y) = p(Y | X)p(X) \tag{1.11}
$$

Here, $p(X, Y)$ is a **joint probability**, verbalized as "the probability of \(X\) and \(Y\)." Similarly, $p(Y | X)$ is a **conditional probability**, verbalized as "the probability of $Y$ given $X$," whereas $p(X)$ is a **marginal probability**, or simply "the probability of $X$."

These two simple rules form the basis for all of the probabilistic machinery used throughout this book.

From the product rule and the symmetry property $p(X, Y) = p(Y, X)$, we can derive the following relationship between conditional probabilities:

$$
p(Y | X) = \frac{p(X | Y) p(Y)}{p(X)} \tag{1.12}
$$

which is **Bayes’ theorem**, playing a central role in pattern recognition and machine learning.

Using the sum rule, the denominator in Bayes’ theorem can be expressed in terms of the quantities appearing in the numerator:

$$
p(X) = \sum_Y p(X | Y)p(Y) \tag{1.13}
$$

The denominator in Bayes’ theorem can be viewed as the normalization constant required to ensure that the sum of the conditional probability on the left-hand side of equation (1.12) over all values of \(Y\) equals one.



Figure.11 An illustration of a distribution over two variables, X, which takes 9 possible values, and Y , which takes two possible values. The top left ﬁgure shows a sample of 60 points drawn from a joint probability distri- bution over these variables. The remaining ﬁgures show histogram estimates of the marginal distributions p(X) and p(Y ), as well as the conditional distribution p(X|Y = 1) corresponding to the bottom row in the top left ﬁgure.

In **Figure.11**, we show a simple example involving a joint distribution over two variables to illustrate the concept of marginal and conditional distributions.

Now, let us return to our example involving boxes of fruit. We have seen that the probabilities of selecting either the red or blue boxes are:

$$
p(B = r) = \frac{4}{10} \tag{1.14}
$$

$$
p(B = b) = \frac{6}{10} \tag{1.15}
$$

These satisfy \(p(B = r) + p(B = b) = 1\).

Now, suppose that we randomly pick a box and it turns out to be the blue box. The probability of selecting an apple is just the fraction of apples in the blue box, which is:

$$
p(F = a | B = b) = \frac{3}{4}
$$

We can write out all four conditional probabilities for the type of fruit given the selected box:

$$
p(F = a | B = r) = \frac{1}{4} \tag{1.16}
$$

$$
p(F = o | B = r) = \frac{3}{4} \tag{1.17}
$$

$$
p(F = a | B = b) = \frac{3}{4} \tag{1.18}
$$


Again, note that these probabilities are normalized so that:

$$
p(F = a | B = r) + p(F = o | B = r) = 1 \tag{1.20}
$$

and similarly:

$$
p(F = a | B = b) + p(F = o | B = b) = 1. \tag{1.21}
$$

We can now use the sum and product rules of probability to evaluate the overall probability of choosing an apple:

$$
p(F = a) = p(F = a | B = r)p(B = r) + p(F = a | B = b)p(B = b) 
$$

$$
= \frac{1}{4} \times \frac{4}{10} + \frac{3}{4} \times \frac{6}{10} = \frac{11}{20}. \tag{1.22}
$$

From which it follows, using the sum rule, that:

$$
p(F = o) = 1 - \frac{11}{20} = \frac{9}{20}.
$$

Suppose instead we are told that a piece of fruit has been selected and it is an orange, and we would like to know which box it came from. This requires that we evaluate the probability distribution over boxes conditioned on the identity of the fruit, whereas the probabilities in (1.16)–(1.19) give the probability distribution over the fruit conditioned on the identity of the box.

We can solve the problem of reversing the conditional probability by using **Bayes’ theorem** to give:

$$
p(B = r | F = o) = \frac{p(F = o | B = r)p(B = r)}{p(F = o)} 
$$

$$
= \frac{\frac{3}{4} \times \frac{4}{10}}{\frac{9}{20}} = \frac{2}{3}. \tag{1.23}
$$

From the sum rule, it follows that:

$$
p(B = b | F = o) = 1 - \frac{2}{3} = \frac{1}{3}.
$$

We can provide an important interpretation of **Bayes’ theorem** as follows. If we had been asked which box had been chosen before being told the identity of the selected item of fruit, then the most complete information we have available is provided by the probability \(p(B)\). We call this the **prior probability** because it is the probability available before we observe the identity of the fruit.

Once we are told that the fruit is an orange, we can then use Bayes’ theorem to compute the probability \(p(B | F)\), which we shall call the **posterior probability** because it is the probability obtained after we have observed \(F\).

In this example, the prior probability of selecting the red box was \(\frac{4}{10}\), so that we were more likely to select the blue box than the red one. However, once we have observed that the piece of selected fruit is an orange, we find that the posterior probability of the red box is now \(\frac{2}{3}\), making it more likely that the box we selected was in fact the red one.

This result accords with our intuition, as the proportion of oranges is much higher in the red box than in the blue box. Thus, the observation that the fruit was an orange provides significant evidence favoring the red box. The evidence is sufficiently strong that it outweighs the prior, making it more likely that the red box was chosen rather than the blue one.

Finally, note that if the joint distribution of two variables factorizes into the product of the marginals, so that:

$$
p(X, Y) = p(X)p(Y),
$$

then \(X\) and \(Y\) are said to be **independent**. From the product rule, we see that:

$$
p(Y | X) = p(Y),
$$

so the conditional distribution of \(Y\) given \(X\) is indeed independent of the value of \(X\).

For instance, in our boxes of fruit example, if each box contained the same fraction of apples and oranges, then \(p(F | B) = p(F)\), meaning the probability of selecting, say, an apple is independent of which box is chosen.


```python
# Given Data
# Red box contains 2 apples, 6 oranges
# Blue box contains 3 apples, 1 orange
# Probability of choosing red box is 4/10, blue box is 6/10

red_box = {"apples": 2, "oranges": 6}
blue_box = {"apples": 3, "oranges": 1}

# Total fruits in red and blue boxes
total_red_box = sum(red_box.values())
total_blue_box = sum(blue_box.values())

# Probability of picking red and blue boxes
p_red_box = 4 / 10
p_blue_box = 6 / 10

# Conditional probabilities: p(F=fruit | B=box)
# Red box
p_apple_given_red = red_box["apples"] / total_red_box  # p(F=a|B=r)
p_orange_given_red = red_box["oranges"] / total_red_box  # p(F=o|B=r)

# Blue box
p_apple_given_blue = blue_box["apples"] / total_blue_box  # p(F=a|B=b)
p_orange_given_blue = blue_box["oranges"] / total_blue_box  # p(F=o|B=b)

# Overall probability of picking an apple (sum rule)
p_apple = (p_apple_given_red * p_red_box) + (p_apple_given_blue * p_blue_box)

# Overall probability of picking an orange (sum rule)
p_orange = 1 - p_apple  # Since only apples and oranges exist, p(F=o) = 1 - p(F=a)

# Now calculate using Bayes' theorem, reversing conditional probabilities
# p(B=r | F=o)
p_orange_given_red = p_orange_given_red
p_orange_given_blue = p_orange_given_blue

# Apply Bayes' Theorem
p_red_given_orange = (p_orange_given_red * p_red_box) / p_orange
p_blue_given_orange = 1 - p_red_given_orange  # Using sum rule

# Results
print(f"Conditional Probabilities:")
print(f"p(apple | red box) = {p_apple_given_red:.2f}")
print(f"p(orange | red box) = {p_orange_given_red:.2f}")
print(f"p(apple | blue box) = {p_apple_given_blue:.2f}")
print(f"p(orange | blue box) = {p_orange_given_blue:.2f}")

print("\nMarginal Probabilities:")
print(f"p(apple) = {p_apple:.2f}")
print(f"p(orange) = {p_orange:.2f}")

print("\nPosterior Probabilities (using Bayes' Theorem):")
print(f"p(red box | orange) = {p_red_given_orange:.2f}")
print(f"p(blue box | orange) = {p_blue_given_orange:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Given Data
red_box = {"apples": 2, "oranges": 6}
blue_box = {"apples": 3, "oranges": 1}

total_red_box = sum(red_box.values())
total_blue_box = sum(blue_box.values())

p_red_box = 4 / 10
p_blue_box = 6 / 10

# Conditional probabilities: p(F=fruit | B=box)
p_apple_given_red = red_box["apples"] / total_red_box  # p(F=a|B=r)
p_orange_given_red = red_box["oranges"] / total_red_box  # p(F=o|B=r)

p_apple_given_blue = blue_box["apples"] / total_blue_box  # p(F=a|B=b)
p_orange_given_blue = blue_box["oranges"] / total_blue_box  # p(F=o|B=b)

# Overall probability of picking an apple (sum rule)
p_apple = (p_apple_given_red * p_red_box) + (p_apple_given_blue * p_blue_box)

# Overall probability of picking an orange (sum rule)
p_orange = 1 - p_apple

# Bayes' theorem to reverse conditional probabilities
p_red_given_orange = (p_orange_given_red * p_red_box) / p_orange
p_blue_given_orange = 1 - p_red_given_orange  # Using sum rule

# Plotting Conditional Probabilities
fruit_types = ["apple", "orange"]
conditional_probabilities = [
    [p_apple_given_red, p_orange_given_red],
    [p_apple_given_blue, p_orange_given_blue]
]
box_labels = ["Red Box", "Blue Box"]

# Create a bar plot for conditional probabilities
plt.figure(figsize=(10, 6))
sns.barplot(x=fruit_types, y=conditional_probabilities[0], label="Red Box", color='red', alpha=0.6)
sns.barplot(x=fruit_types, y=conditional_probabilities[1], label="Blue Box", color='blue', alpha=0.6)

plt.title("Conditional Probabilities of Picking a Fruit from Each Box")
plt.ylabel("Probability")
plt.legend(title="Boxes")
plt.show()

# Plotting Posterior Probabilities (Bayes' Theorem)
posterior_probabilities = [p_red_given_orange, 1 - p_red_given_orange]
box_labels = ["Red Box", "Blue Box"]

plt.figure(figsize=(8, 6))
sns.barplot(x=box_labels, y=posterior_probabilities, palette=["red", "blue"])

plt.title("Posterior Probabilities (Bayes' Theorem) \n Given that an Orange was Selected")
plt.ylabel("Probability")
plt.show()

# Plotting Marginal Probabilities
marginal_probabilities = [p_apple, p_orange]

plt.figure(figsize=(8, 6))
sns.barplot(x=fruit_types, y=marginal_probabilities, palette=["green", "orange"])

plt.title("Marginal Probabilities of Selecting an Apple or Orange")
plt.ylabel("Probability")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Given Data
red_box = {"apples": 2, "oranges": 6}
blue_box = {"apples": 3, "oranges": 1}

total_red_box = sum(red_box.values())
total_blue_box = sum(blue_box.values())

p_red_box = 4 / 10
p_blue_box = 6 / 10

# Conditional probabilities: p(F=fruit | B=box)
p_apple_given_red = red_box["apples"] / total_red_box  # p(F=a|B=r)
p_orange_given_red = red_box["oranges"] / total_red_box  # p(F=o|B=r)

p_apple_given_blue = blue_box["apples"] / total_blue_box  # p(F=a|B=b)
p_orange_given_blue = blue_box["oranges"] / total_blue_box  # p(F=o|B=b)

# Overall probability of picking an apple and orange (sum rule)
p_apple = (p_apple_given_red * p_red_box) + (p_apple_given_blue * p_blue_box)
p_orange = 1 - p_apple

# Bayes' theorem to reverse conditional probabilities
p_red_given_orange = (p_orange_given_red * p_red_box) / p_orange
p_blue_given_orange = 1 - p_red_given_orange  # Using sum rule

# Set up the figure and axis for animation
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Function to initialize the plots
def init():
    for ax in axs:
        ax.clear()
    return axs

# Animation function
def animate(i):
    axs[0].clear()
    axs[1].clear()
    axs[2].clear()

    # Update data for conditional probabilities
    if i < 10:
        scale = (i + 1) / 10  # Scale factor for animation
    else:
        scale = 1  # Max scale

    # Conditional probabilities
    conditional_probabilities = [
        [p_apple_given_red * scale, p_orange_given_red * scale],
        [p_apple_given_blue * scale, p_orange_given_blue * scale]
    ]
    fruit_types = ["Apple", "Orange"]
    box_labels = ["Red Box", "Blue Box"]

    # Create a bar plot for conditional probabilities
    sns.barplot(x=fruit_types, y=conditional_probabilities[0], label="Red Box", color='red', alpha=0.6, ax=axs[0])
    sns.barplot(x=fruit_types, y=conditional_probabilities[1], label="Blue Box", color='blue', alpha=0.6, ax=axs[0])
    axs[0].set_title("Conditional Probabilities of Picking a Fruit from Each Box")
    axs[0].set_ylabel("Probability")
    axs[0].legend(title="Boxes")

    # Posterior probabilities (Bayes' Theorem)
    posterior_probabilities = [p_red_given_orange * scale, p_blue_given_orange * scale]
    sns.barplot(x=box_labels, y=posterior_probabilities, palette=["red", "blue"], ax=axs[1])
    axs[1].set_title("Posterior Probabilities Given that an Orange was Selected")
    axs[1].set_ylabel("Probability")

    # Marginal probabilities
    marginal_probabilities = [p_apple * scale, p_orange * scale]
    sns.barplot(x=["Apple", "Orange"], y=marginal_probabilities, palette=["green", "orange"], ax=axs[2])
    axs[2].set_title("Marginal Probabilities of Selecting an Apple or Orange")
    axs[2].set_ylabel("Probability")

# Create animation
ani = FuncAnimation(fig, animate, frames=20, init_func=init, repeat=False)

# Show the animation
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Given Data
box_a = {"red": 4, "blue": 6}
box_b = {"red": 3, "blue": 7}

total_a = sum(box_a.values())
total_b = sum(box_b.values())

# Prior probabilities of choosing each box
p_box_a = 0.5
p_box_b = 0.5

# Initialize figures
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Function to initialize the plots
def init():
    for ax in axs:
        ax.clear()
    return axs

# Animation function
def animate(i):
    axs[0].clear()
    axs[1].clear()
    axs[2].clear()

    # Update the data for animation
    scale = min(i + 1, 10) / 10  # Scale factor for animation

    # Conditional probabilities
    p_red_given_a = (box_a["red"] / total_a) * scale
    p_blue_given_a = (box_a["blue"] / total_a) * scale
    p_red_given_b = (box_b["red"] / total_b) * scale
    p_blue_given_b = (box_b["blue"] / total_b) * scale

    # Create a bar plot for conditional probabilities
    sns.barplot(x=["Red (A)", "Blue (A)", "Red (B)", "Blue (B)"], 
                y=[p_red_given_a, p_blue_given_a, p_red_given_b, p_blue_given_b], 
                palette=["red", "blue", "red", "blue"], ax=axs[0])
    axs[0].set_title("Conditional Probabilities of Picking a Ball from Each Box")
    axs[0].set_ylabel("Probability")
    axs[0].set_ylim(0, 1)

    # Overall probability of picking a red or blue ball (sum rule)
    p_red = (p_red_given_a * p_box_a) + (p_red_given_b * p_box_b)
    p_blue = (p_blue_given_a * p_box_a) + (p_blue_given_b * p_box_b)

    # Create a bar plot for overall probabilities
    sns.barplot(x=["Red", "Blue"], y=[p_red * scale, p_blue * scale], palette=["red", "blue"], ax=axs[1])
    axs[1].set_title("Overall Probabilities of Picking a Ball")
    axs[1].set_ylabel("Probability")
    axs[1].set_ylim(0, 1)

    # Update posterior probabilities after observing a red ball
    if i > 0:
        p_a_given_red = (p_red_given_a * p_box_a) / p_red
        p_b_given_red = (p_red_given_b * p_box_b) / p_red
    else:
        p_a_given_red = p_box_a
        p_b_given_red = p_box_b

    # Create a bar plot for posterior probabilities
    sns.barplot(x=["Box A", "Box B"], y=[p_a_given_red * scale, p_b_given_red * scale], palette=["red", "blue"], ax=axs[2])
    axs[2].set_title("Posterior Probabilities Given a Red Ball was Picked")
    axs[2].set_ylabel("Probability")
    axs[2].set_ylim(0, 1)

# Create animation
ani = FuncAnimation(fig, animate, frames=20, init_func=init, repeat=False)

# Show the animation
plt.tight_layout()
plt.show()
```
