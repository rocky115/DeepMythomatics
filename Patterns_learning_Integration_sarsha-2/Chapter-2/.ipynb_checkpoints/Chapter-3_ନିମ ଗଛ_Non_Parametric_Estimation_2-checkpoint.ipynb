##  The Nearest-Neighbor Rule

While the $k $-nearest-neighbor algorithm was first proposed for arbitrary $ k $, the crucial matter of determining the error bound was first solved for $ k = 1 $. This nearest-neighbor algorithm has conceptual and computational simplicity.

##  Convergence of the Nearest Neighbor

We now wish to evaluate the average probability of error for the nearest-neighbor rule. In particular, if $ P_n(e) $ is the $ n $-sample error rate, and if 

$$
P = \lim_{n \to \infty} P_n(e),
$$ 

then we want to show that 

$$
c P^* \leq P \leq \frac{P^*}{2 - P^*}.
$$ 

We begin by observing that when the nearest-neighbor rule is used with a particular set of $ n $ samples, the resulting error rate will depend on the accidental characteristics of the samples. In particular, if different sets of $ n $ samples are used to classify $ x $, different vectors $ x^* $ will be obtained for the nearest-neighbor of $ x $. Since the decision rule depends on this nearest-neighbor, we have a conditional probability of error $ P(e|x, x^*) $ that depends on both $ x $ and $ x^* $. By averaging over $ x^* $, we obtain 

$$
P(e|x) = \int P(e|x, x^*) p(x^*|x) \, dx^*.
$$ 

where we understand that there is an implicit dependence upon the number $ n $ of training points. It is usually very difficult to obtain an exact expression for the conditional density $ p(x^*|x) $. However, since $ x^* $ is by definition the nearest-neighbor of $ x $, we expect this density to be very peaked in the immediate vicinity of $ x $, and very small elsewhere. Furthermore, as $ n $ goes to infinity we expect $ p(x^*|x) $ to approach a delta function centered at $ x $, making the evaluation of Eq. (41) trivial. 

To show that this is indeed the case, we must assume that at the given $ x $, $ p(\cdot) $ is continuous and not equal to zero. Under these conditions, the probability that any sample falls within a hypersphere $ S $ centered about $ x $ is some positive number $ P_s $:

$$
P_s = \int_{x' \in S} p(x') \, dx'.
$$ 

Thus, the probability that all $ n $ of the independently drawn samples fall outside this hypersphere is $ (1 - P_s)^n $, which approaches zero as $ n $ goes to infinity. Thus $ x^* $ converges to $ x $ in probability, and $ p(x^*|x) $ approaches a delta function, as expected. In fact, by using measure theoretic methods one can make even stronger (as well as more rigorous) statements about the convergence of $ x^* $ to $ x $, but this result is sufficient for our purposes.

##  Error Rate for the Nearest-Neighbor Rule

We now turn to the calculation of the conditional probability of error $ P_n(e|x, x^*_j) $. To avoid a potential source of confusion, we must state the problem with somewhat greater care than has been exercised so far. When we say that we have $ n $ independently drawn labelled samples, we are talking about $ n $ pairs of random variables $ (x_1, \theta_1), (x_2, \theta_2), \ldots, (x_n, \theta_n) $, where $ \theta_j $ may be any of the $ c $ states of nature $ \omega_1, \ldots, \omega_c $. We assume that these pairs were generated by selecting a state of nature $ \omega_j $ for $ \theta_j $ with probability $ P(\omega_j) $ and then selecting an $ x_j $ according to the probability law $ p(x|\omega_j) $, with each pair being selected independently. 

Suppose that during classification nature selects a pair $ (x, \theta) $, and that $ x^*_j $, labelled $ \theta_{j^*} $, is the training sample nearest $ x $. Since the state of nature when $ x^*_j $ was drawn is independent of the state of nature when $ x $ is drawn, we have 

$$
P(\theta, \theta_{j^*} | x, x^*_j) = P(\theta | x) P(\theta_{j^*} | x^*_j).
$$ 

Now if we use the nearest-neighbor decision rule, we commit an error whenever $ \theta = \theta_{j^*} $. Thus, the conditional probability of error $ P_n(e|x, x^*_j) $ is given by 

$$
P_n(e|x, x^*_j) = 1 - \sum_{i=1}^{c} P(\theta = \omega_i, \theta_{j^*} = \omega_i | x, x^*_j) = 1 - \sum_{i=1}^{c} P(\omega_i | x) P(\omega_i | x^*_j).
$$ 

To obtain $ P_n(e) $ we must substitute this expression into Eq. (41) for $ P_n(e|x) $ and then average the result over $ x $. This is very difficult, in general, but as we remarked earlier the integration called for in Eq. (41) becomes trivial as $ n $ goes to infinity and $ p(x^*|x) $ approaches a delta function. If $ P(\omega_i | x) $ is continuous at $ x $, we thus obtain 

$$
\lim_{n \to \infty} P_n(e|x) = 1 - \sum_{i=1}^{c} P(\omega_i | x) P(\omega_i | x^*) \delta(x^* - x) \, dx^* = 1 - \sum_{i=1}^{c} P^2(\omega_i | x).
$$ 

Therefore, provided we can exchange some limits and integrals, the asymptotic nearest-neighbor error rate is given by 

$$
P = \lim_{n \to \infty} P_n(e) = \lim_{n \to \infty} P_n(e|x) p(x) \, dx = 1 - \sum_{i=1}^{c} P^2(\omega_i | x) p(x) \, dx.
$$ 



```python
# kNN Implementation for the Fish Dataset

import random
import seaborn as sns
import matplotlib.pyplot as plt

# Define the Fish dataset
def load_fish_data():
    # Sample data: features are lengths and weights; labels are species
    data = [
        (30, 300, 'Bream'),
        (35, 400, 'Bream'),
        (32, 350, 'Bream'),
        (50, 500, 'Perch'),
        (55, 600, 'Perch'),
        (52, 550, 'Perch'),
        (20, 200, 'Smelt'),
        (22, 220, 'Smelt'),
        (21, 210, 'Smelt')
    ]
    return data

# Calculate Euclidean distance
def euclidean_distance(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

# k-NN Classifier
class KNN:
    def __init__(self, k=3):
        self.k = k
        self.data = []

    def fit(self, data):
        self.data = data

    def predict(self, new_point):
        distances = []
        for point in self.data:
            dist = euclidean_distance(new_point[:2], point[:2])
            distances.append((dist, point[2]))  # (distance, label)

        distances.sort(key=lambda x: x[0])
        nearest_neighbors = [label for _, label in distances[:self.k]]

        return max(set(nearest_neighbors), key=nearest_neighbors.count)

# Function to visualize the data and decision boundaries
def plot_decision_boundaries(knn, data):
    # Create a mapping of species to integers
    species_mapping = {species: idx for idx, species in enumerate(set(d[2] for d in data))}
    
    # Create a mesh to plot
    x_min, x_max = 10, 70
    y_min, y_max = 100, 700
    xx = list(range(x_min, x_max))  # x-coordinates
    yy = list(range(y_min, y_max))   # y-coordinates

    # Create a grid for predictions
    Z = [[0 for _ in range(len(xx))] for _ in range(len(yy))]
    
    for i, y in enumerate(yy):
        for j, x in enumerate(xx):
            pred = knn.predict((x, y, None))
            Z[i][j] = species_mapping[pred]  # Convert prediction to mapped integer

    # Plotting the decision boundaries
    plt.contourf(xx, yy, Z, alpha=0.5, cmap='coolwarm')
    sns.scatterplot(x=[d[0] for d in data], y=[d[1] for d in data], hue=[d[2] for d in data], palette='Set1', edgecolor='w')
    plt.title('k-NN Decision Boundaries')
    plt.xlabel('Length')
    plt.ylabel('Weight')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.show()

# Main function to execute the k-NN algorithm
def main():
    # Load data
    data = load_fish_data()
    
    # Instantiate and train the k-NN classifier
    knn = KNN(k=3)
    knn.fit(data)

    # Predict a new point (length, weight)
    new_point = (40, 450, None)  # Example input point
    prediction = knn.predict(new_point)
    print(f'The predicted species for the point {new_point[:2]} is: {prediction}')

    # Plot decision boundaries
    plot_decision_boundaries(knn, data)

# Run the main function
if __name__ == '__main__':
    main()
```

##  Error Bounds

While Eq. 46 presents an exact result, it is more illuminating to obtain bounds on $ P $ in terms of the Bayes rate $ P^* $. An obvious lower bound on $ P $ is $ P^* $ itself. Furthermore, it can be shown that for any $ P^* $ there is a set of conditional and prior probabilities for which the bound is achieved, so in this sense it is a tight lower bound.

The problem of establishing a tight upper bound is more interesting. The basis for hoping for a low upper bound comes from observing that if the Bayes rate is low, $ P(\omega_i | x) $ is near $ 1.0 $ for some $ i $, say $ i = m $. Thus, the integrand in Eq. 46 is approximately

$$
1 - P^2(\omega_m | x) \approx 2(1 - P(\omega_m | x)),
$$

and since $ P^*(e|x) = 1 - P(\omega_m | x) $, integration over $ x $ might yield about twice the Bayes rate, which is still low and acceptable for some applications.

To obtain an exact upper bound, we must find out how large the nearest-neighbor error rate $ P $ can become for a given Bayes rate $ P^* $. Thus, Eq. 46 leads us to ask how small $ P^2(\omega_i | x) $ can be for a given $ P(\omega_m | x) $. 

First, we write

$$
P^2(\omega_i | x) = P^2(\omega_m | x) + P^2(\omega_i | x) \quad \text{for } i \neq m,
$$

and then seek to bound this sum by minimizing the second term subject to the following constraints:

- $ P(\omega_i | x) \geq 0 $
- $ P(\omega_i | x) = 1 - P(\omega_m | x) = P^*(e|x) \quad \text{for } i = m. $

With a little thought, we see that $ P^2(\omega_i | x) $ is minimized if all of the posterior probabilities except the $ m $th are equal. The second constraint yields

$$
P^*(e|x) = \frac{c-a}{1 - P^*(e|x)} \quad \text{for } i \neq m.
$$

Thus we have the inequalities

$$
\sum_{i=1}^{c} P^2(\omega_i | x) \geq (1 - P^*(e|x))^2 + \sum_{i \neq m}^{c-1} P^2(\omega_i | x), 
$$

and 

$$
\sum_{i=1}^{c} (1 - P^2(\omega_i | x)) \leq 2P^*(e|x) - P^{*2}(e|x).
$$

This immediately shows that $ P \leq 2P^* $, since we can substitute this result in Eq. 46 and merely drop the second term.

However, a tighter bound can be obtained by observing that the variance is:

$$
\text{Var}[P^*(e|x)] = \int [P^*(e|x) - P^*]^2 p(x) \, dx = \int P^{*2}(e|x)p(x) \, dx - P^{*2} \geq 0,
$$

so that 

$$
\int P^{*2}(e|x)p(x) \, dx \geq P^{*2},
$$

with equality holding if and only if the variance of $ P^*(e|x) $ is zero. Using this result and substituting the previous inequality into Eq. 46, we obtain the desired bounds on the nearest-neighbor error $ P $ in the case of an infinite number of samples:

$$
P^* \leq P \leq \frac{P^{*2}}{c-1}.
$$

It is easy to show that this upper bound is achieved in the so-called zero-information case in which the densities $ p(x|\omega_i) $ are identical, so that $ P(\omega_i | x) = P(\omega_i) $ and furthermore $ P^*(e|x) $ is independent of $ x $ (Problem 17). 

Thus the bounds given by Eq. 53 are as tight as possible, in the sense that for any $ P^* $ there exist conditional and a priori probabilities for which the bounds are achieved. In particular, the Bayes rate $ P^* $ can be anywhere between $ 0 $ and $ \frac{c-1}{c} $ and the bounds meet at the two extreme values for the probabilities. 

When the Bayes rate is small, the upper bound is approximately twice the Bayes rate (Fig.14).

$$
P \quad \text{for} \quad c-1 \quad \text{to} \quad c
$$

$$
P^* \quad \text{for} \quad c-1 \quad \text{to} \quad c
$$

At low error rates, the nearest-neighbor error rate is bounded above by twice the Bayes rate. Since $ P $ is always less than or equal to $ 2P^* $, if one had an infinite collection of data and used an arbitrarily complicated decision rule, one could at most cut the error rate in half. In this sense, at least half of the classification information in an infinite dataset resides in the nearest neighbor.

It is natural to ask how well the nearest-neighbor rule works in the finite-sample case, and how rapidly the performance converges to the asymptotic value. Unfortunately, despite prolonged effort on such problems, the only statements that can be made in the general case are negative. It can be shown that convergence can be arbitrarily slow, and the error rate $ P_n(e) $ need not even decrease monotonically with $ n $. As with other nonparametric methods, it is difficult to obtain anything other than asymptotic results without making further assumptions about the underlying probability structure (Problems 13 & 14).


```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to generate a synthetic dataset
def generate_data(num_samples=100):
    # Generate random data points
    data = []
    for _ in range(num_samples):
        # Randomly assign class labels (0 or 1)
        label = random.choice([0, 1])
        x1 = random.uniform(0, 10)
        x2 = random.uniform(0, 10)
        data.append((x1, x2, label))
    return data

# Function to calculate the nearest neighbor classification
def knn_predict(data, query, k=1):
    # Calculate distances from the query point to all points in the data
    distances = []
    for (x1, x2, label) in data:
        distance = ((x1 - query[0]) ** 2 + (x2 - query[1]) ** 2) ** 0.5
        distances.append((distance, label))
    
    # Sort by distance and take the top k
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    
    # Vote for the class
    class_votes = {}
    for _, label in neighbors:
        class_votes[label] = class_votes.get(label, 0) + 1
    return max(class_votes, key=class_votes.get)

# Function to calculate the error rate of the k-NN classifier
def calculate_error(data, k=1):
    errors = 0
    for point in data:
        # Separate the features and the label
        query = point[:2]  # (x1, x2)
        true_label = point[2]
        
        # Predict the label
        predicted_label = knn_predict(data, query, k)
        
        # Count the error
        if predicted_label != true_label:
            errors += 1
    return errors / len(data)

# Function to plot decision boundaries
def plot_decision_boundaries(data, k=1):
    # Create a grid to plot the decision boundaries
    x_min, x_max = 0, 10
    y_min, y_max = 0, 10
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

    # Predict for each point in the grid
    Z = []
    for i in range(xx.shape[0]):
        row = []
        for j in range(xx.shape[1]):
            query = (xx[i, j], yy[i, j])
            predicted_label = knn_predict(data, query, k)
            row.append(predicted_label)
        Z.append(row)
    
    Z = np.array(Z)

    # Plotting the decision boundaries
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.5, cmap='coolwarm')
    sns.scatterplot(x=[d[0] for d in data], y=[d[1] for d in data], hue=[d[2] for d in data], palette='Set1', edgecolor='w')
    plt.title('k-NN Decision Boundaries')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Main function to execute the k-NN algorithm and plot the results
def main():
    # Generate synthetic data
    data = generate_data(200)

    # Calculate error rate
    error_rate = calculate_error(data, k=3)
    print(f'Error Rate: {error_rate:.2f}')

    # Plot decision boundaries
    plot_decision_boundaries(data, k=3)

# Run the main function
if __name__ == '__main__':
    main()
```



## The k-Nearest-Neighbor Rule

An obvious extension of the nearest-neighbor rule is the **k-nearest-neighbor rule**. As one would expect from the name, this rule classifies $x$ by assigning it the label most frequently represented among the $k$ nearest samples. In other words, a decision is made by examining the labels on the $k$ nearest neighbors and taking a vote (Fig.15).




  
*Fig.15: The k-nearest-neighbor query starts at the test point and grows a spherical region until it encloses $k$ training samples, and labels the test point by a majority vote of these samples. In this $k = 5$ case, the test point $x$ would be labelled the category of the black points.*

The basic motivation for considering the k-nearest-neighbor rule rests on our earlier observation about matching probabilities with nature. We notice first that if $k$ is fixed and the number $n$ of samples is allowed to approach infinity, then all of the $k$ nearest neighbors will converge to $x$. Hence, as in the single-nearest-neighbor cases, the labels on each of the $k$-nearest-neighbors are random variables, which independently assume the values $\omega_i$ with probabilities $P(\omega_i | x)$, $i = 1, 2$. If $P(\omega_m | x)$ is the larger a posteriori probability, then the Bayes decision rule always selects $\omega_m$.

The single-nearest-neighbor rule selects $\omega_m$ with probability $P(\omega_m | x)$. The k-nearest-neighbor rule selects $\omega_m$ if a majority of the $k$ nearest neighbors are labeled $\omega_m$, an event of probability:

$$
P\left(\text{select } \omega_m\right) = \sum_{i=\frac{k+1}{2}}^{k} \binom{k}{i} P(\omega_m | x)^i [1 - P(\omega_m | x)]^{k-i}
$$

In general, the larger the value of $k$, the greater the probability that $\omega_m$ will be selected.

We could analyze the k-nearest-neighbor rule in much the same way that we analyzed the single-nearest-neighbor rule. However, since the arguments become more involved and supply little additional insight, we shall content ourselves with stating the results.

It can be shown that if $k$ is odd, the large-sample two-class error rate for the k-nearest-neighbor rule is bounded above by the function $C_k(P^*)$, where $C_k(P^*)$ is defined to be the smallest concave function of $P^*$ greater than:

$$
\frac{k-1}{2} \sum_{i=0}^{k} \binom{k}{i} (P^*)^{i+1} (1 - P^*)^{k-i} + (P^*)^{k-i} (1 - P^*)^{i+1}.
$$

Here the summation over the first bracketed term represents the probability of error due to $i$ points coming from the category having the minimum probability and $k-i > i$ points from the other category. The summation over the second term in the brackets is the probability that $k - i$ points are from the minimum-probability category and $i + 1 < k - i$ from the higher probability category. Both of these cases constitute errors under the k-nearest-neighbor decision rule, and thus we must add them to find the full probability of error.



*Fig.16: The error-rate for the k-nearest-neighbor rule for a two-category problem is bounded by $C_k(P^*)$ in Eq. 55. Each curve is labelled by $k$; when $k = \infty$, the estimated probabilities match the true probabilities and thus the error rate is equal to the Bayes rate, i.e., $P = P^*$.*

At the risk of sounding repetitive, we conclude by commenting once again on the finite-sample situation encountered in practice. The k-nearest-neighbor rule can be viewed as another attempt to estimate the a posteriori probabilities $P(\omega_i | x)$ from samples. We want to use a large value of $k$ to obtain a reliable estimate. On the other hand, we want all of the $k$ nearest neighbors $x_{k}$ to be very near $x$ to be sure that $P(\omega_i | x_{k})$ is approximately the same as $P(\omega_i | x)$. This forces us to choose a compromise $k$ that is a small fraction of the number of samples. It is only in the limit as $n$ goes to infinity that we can be assured of the nearly optimal behavior of the k-nearest-neighbor rule.


```python
import matplotlib.pyplot as plt
import random

# Generate synthetic data for demonstration
def generate_data(n_samples=100, n_classes=2, random_state=42):
    random.seed(random_state)
    data = []
    for i in range(n_classes):
        class_data = [(random.uniform(0, 10), random.uniform(0, 10), i) for _ in range(n_samples)]
        data.extend(class_data)
    return data

# Compute the Euclidean distance
def euclidean_distance(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

# k-NN classification
def k_nearest_neighbors(data, new_point, k=3):
    # Sort the data by distance to the new point
    distances = [(euclidean_distance((point[0], point[1]), new_point), point[2]) for point in data]
    distances.sort(key=lambda x: x[0])
    
    # Get the labels of the k nearest neighbors
    k_nearest = distances[:k]
    labels = [label for _, label in k_nearest]
    
    # Return the most common label
    return max(set(labels), key=labels.count)

# Plot decision boundaries
def plot_decision_boundaries(data, k=3):
    # Define grid
    x_min, x_max = 0, 10
    y_min, y_max = 0, 10
    step_size = 0.1  # Step size for the grid
    xx, yy = [], []

    # Create grid points
    for x in range(int(x_min * 10), int(x_max * 10) + 1):
        for y in range(int(y_min * 10), int(y_max * 10) + 1):
            xx.append(x / 10.0)
            yy.append(y / 10.0)
    
    # Predict for each point in the grid
    Z = []
    for i in range(len(xx)):
        new_point = (xx[i], yy[i])
        label = k_nearest_neighbors(data, new_point, k)
        Z.append(label)

    # Convert Z to a 2D grid for contour plotting
    Z = [[Z[i + j * int((x_max - x_min) / step_size + 1)] for i in range(int((x_max - x_min) / step_size + 1)] 
          for j in range(int((y_max - y_min) / step_size + 1)]

    # Plot decision boundaries
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')

    # Plot original data points
    for point in data:
        plt.scatter(point[0], point[1], c='black' if point[2] == 0 else 'red')

    plt.title(f'k-NN Decision Boundaries (k={k})')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.show()

# Main function
def main():
    # Generate synthetic data
    data = generate_data(n_samples=100, n_classes=2, random_state=42)
    
    # Plot decision boundaries
    plot_decision_boundaries(data, k=3)

# Run the main function
if __name__ == '__main__':
    main()
```
