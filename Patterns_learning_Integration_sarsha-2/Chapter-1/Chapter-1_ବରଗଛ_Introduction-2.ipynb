## Bayesian Decision Theory

## 1 Introduction

Bayesian decision theory is a fundamental statistical approach to the problem of pattern classification. This approach is based on quantifying the trade-offs between various classification decisions using probability and the costs associated with them. It assumes that the decision problem is framed probabilistically and that all relevant probabilities are known. 

In this chapter, we develop the fundamentals of Bayesian decision theory, starting with a specific example before moving into general formalizations. Later, we will discuss issues that arise when the probabilistic structure is not entirely known.

### Example: Classifying Fish as Sea Bass or Salmon

Consider the hypothetical problem of designing a classifier to separate two kinds of fish: sea bass and salmon. Suppose an observer watching fish arrive on a conveyor belt finds it difficult to predict what type will appear next. In decision-theoretic terminology, we would say that each fish has a random state: either the fish is a sea bass ($ \omega = \omega_1 $) or a salmon ($ \omega = \omega_2 $).

Let:
- $ P(\omega_1) $: Prior probability of the fish being a sea bass
- $ P(\omega_2) $: Prior probability of the fish being a salmon

Since we only consider two fish types:
$$
P(\omega_1) + P(\omega_2) = 1
$$

If we were forced to make a decision based only on the priors, we would decide:
- $ \omega_1 $ if $ P(\omega_1) > P(\omega_2) $
- $ \omega_2 $ otherwise

This rule works well if the prior probabilities are known, but if we had additional information, such as a lightness measurement \( x \) for each fish, we could improve the classification.

## Conditional Densities and Bayes' Formula

The lightness measurement $ x $ can be expressed as a continuous random variable whose distribution depends on the state of nature $( \omega $):
- $ p(x|\omega_1) $: Probability density function of $ x $ given that the fish is sea bass
- $ p(x|\omega_2) $: Probability density function of $ x $ given that the fish is salmon

If we measure the lightness of a fish and observe a specific value $ x $, we can use Bayes' formula to update our beliefs about the fish's category.

### Bayes' Formula

The joint probability of observing a fish in category $ \omega_j $ with feature $ x $ can be expressed as:
$
p(\omega_j, x) = P(\omega_j | x) p(x) = p(x | \omega_j) P(\omega_j)
$

Rearranging this, we get Bayes' formula:
$
P(\omega_j | x) = \frac{p(x | \omega_j) P(\omega_j)}{p(x)}
$
where
$
p(x) = \sum_{j=1}^{2} p(x | \omega_j) P(\omega_j)
$

#### In words:
$$
\text{posterior} = \frac{\text{likelihood} \times \text{prior}}{\text{evidence}}
$$

Bayes' formula allows us to convert the prior probability $ P(\omega_j) $ into the posterior probability $ P(\omega_j | x) $, or the probability that the fish is in category $ \omega_j $ given the lightness $ x $.

## Decision Rule Based on Posterior Probabilities

To minimize the probability of error:
- Decide $ \omega_1 $ if $ P(\omega_1 | x) > P(\omega_2 | x) $
- Otherwise, decide $ \omega_2 $

For a specific observation $ x $, the probability of error is given by:
$$
P(\text{error} | x) = 
\begin{cases} 
P(\omega_1 | x) & \text{if we decide } \omega_2 \\
P(\omega_2 | x) & \text{if we decide } \omega_1 
\end{cases}
$$

To minimize the **average probability of error**, we use the rule that minimizes $ P(\text{error} | x) $ for each $ x $. This approach, as shown above, is optimal in reducing the probability of error for Bayesian decision-making.

### Posterior Probabilities and Error Minimization

In this section, we illustrate how posterior probabilities help in making optimal decisions under uncertainty. Consider the following posterior probabilities for the priors $ P(\omega_1) = \frac{2}{3} $ and $ P(\omega_2) = \frac{1}{3} $, based on class-conditional probability densities.

#### Posterior Probability Graph (Fig.2)

This graph shows the posterior probabilities for a feature value $ x $. In this example, given that a pattern is measured to have a feature value of $ x = 14 $:
- The probability it belongs to category $ \omega_2 $ is roughly $ 0.08 $
- The probability it belongs to category $ \omega_1 $ is approximately $ 0.92 $

At each $ x $, the posterior probabilities sum to $ 1 $, as required:
$$
P(\omega_1 | x) + P(\omega_2 | x) = 1
$$

#### Calculating the Probability of Error

The probability of error, $ P(\text{error}) $, can be calculated as follows:
$$
P(\text{error}) = \int_{-\infty}^{\infty} P(\text{error}, x) \, dx = \int_{-\infty}^{\infty} P(\text{error} | x) \, p(x) \, dx
$$

To minimize $ P(\text{error}) $, we ensure that $ P(\text{error} | x) $ is minimized for every $ x $. This leads to the Bayes decision rule:

### Bayes' Decision Rule for Minimizing Error

To minimize the probability of error, we use the following rule:
$$
\text{Decide } \omega_1 \text{ if } P(\omega_1 | x) > P(\omega_2 | x); \text{ otherwise decide } \omega_2
$$

This rule minimizes $ P(\text{error} | x) $ by choosing the category with the higher posterior probability:

$$
P(\text{error} | x) = \min \left[ P(\omega_1 | x), P(\omega_2 | x) \right]
$$

### Decision Rule in Terms of Conditional and Prior Probabilities

Using Bayes' formula (Eq. 1), we can express the decision rule without the evidence term $ p(x) $ since it does not affect the decision. This gives us an equivalent decision rule:
$$
\text{Decide } \omega_1 \text{ if } p(x | \omega_1) P(\omega_1) > p(x | \omega_2) P(\omega_2); \text{ otherwise decide } \omega_2
$$

This form emphasizes the importance of both the likelihoods $ p(x | \omega_j) $ and the prior probabilities $ P(\omega_j) $ in making the decision.

#### Special Cases

- **If $ P(\omega_1) \gg P(\omega_2) $**: The decision heavily favors $ \omega_1 $ due to the prior probability, regardless of the likelihoods.
- **If $ P(\omega_1) = P(\omega_2) $**: The decision is based entirely on the likelihoods $ p(x | \omega_j) $.

In general, the Bayes decision rule optimally combines the prior probabilities and likelihoods to minimize the probability of error, resulting in a statistically optimal decision-making process.

### Bayesian Decision Theory – Continuous Features

We now formalize Bayesian decision theory, generalizing it in several ways:
- Allowing more than one feature (vector representation).
- Considering more than two states of nature.
- Permitting actions beyond just deciding the state of nature (e.g., rejection).
- Introducing a general loss function, extending beyond the probability of error.

These extensions add complexity but preserve the core ideas. Let’s go through each in turn.

#### Generalization of Features

When we have multiple features, we represent them as a feature vector $ \mathbf{x} $ in a $ d $-dimensional Euclidean space $ \mathbb{R}^d $, called the **feature space**.

#### Generalization to Multiple States of Nature

Instead of two states, we consider a set of $ c $ states of nature (or categories), denoted $ \omega_1, \omega_2, \ldots, \omega_c $.

#### Generalization to Multiple Actions

We allow a set of $ a $ possible actions, denoted $ \alpha_1, \alpha_2, \ldots, \alpha_a $. Additional actions allow for choices such as **rejection** in cases where decision confidence is low, useful if indecision is not prohibitively costly.

#### Introduction of a Loss Function

The **loss function**, $ \lambda(\alpha_i | \omega_j) $, defines the cost of taking action $ \alpha_i $ when the true state of nature is $ \omega_j $. This allows us to assign different weights to types of errors, rather than treating all misclassifications equally.

### Mathematical Framework

Given these generalizations, we proceed with the following definitions and formulations.

Let:
- $\mathbf{x} $ be a $ d $-dimensional feature vector, representing observations.
- $ p(\mathbf{x} | \omega_j) $ be the **class-conditional probability density function** for $ \mathbf{x} $ given state $ \omega_j $.
- $ P(\omega_j) $denote the **prior probability** that the state of nature is $\omega_j $

Using Bayes' theorem, the **posterior probability** \( P(\omega_j | \mathbf{x}) \) is:
$$
P(\omega_j | \mathbf{x}) = \frac{p(\mathbf{x} | \omega_j) P(\omega_j)}{p(\mathbf{x})}
$$
where the **evidence** \( p(\mathbf{x}) \) is:
$$
p(\mathbf{x}) = \sum_{j=1}^{c} p(\mathbf{x} | \omega_j) P(\omega_j)
$$

### Expected Loss and Conditional Risk

Suppose we observe $ \mathbf{x} $ and consider taking action $ \alpha_i $. If the true state of nature is $ \omega_j $, the loss incurred is $ \lambda(\alpha_i | \omega_j) $. Given that $ P(\omega_j | \mathbf{x}) $ is the probability of the true state being $ \omega_j $, the **expected loss** (or **conditional risk**) associated with action $ \alpha_i $ is:
$$
R(\alpha_i | \mathbf{x}) = \sum_{j=1}^{c} \lambda(\alpha_i | \omega_j) P(\omega_j | \mathbf{x})
$$

In decision theory, this expected loss is termed **risk**, and $ R(\alpha_i | \mathbf{x}) $ is the **conditional risk**.

### Bayes Decision Procedure

To minimize the **overall risk**, we select the action that minimizes the conditional risk for each observation $ \mathbf{x} $. Formally, the **Bayes decision rule** states:

- Compute the conditional risk $ R(\alpha_i | \mathbf{x}) $ for each action $ \alpha_i $:
$$
R(\alpha_i | \mathbf{x}) = \sum_{j=1}^{c} \lambda(\alpha_i | \omega_j) P(\omega_j | \mathbf{x})
$$
- Choose the action $ \alpha_i $ that minimizes $ R(\alpha_i | \mathbf{x}) $.

The resulting minimum overall risk is called the **Bayes risk**, denoted $ R^* $, and represents the best achievable performance.

### Overall Risk

The **overall risk**, $ R $, associated with a decision rule is the expected loss over all possible observations, given by:
$$
R = \int R(\alpha(\mathbf{x}) | \mathbf{x}) p(\mathbf{x}) \, d\mathbf{x}
$$
where $ d\mathbf{x} $ denotes a volume element in $ d $-space, and the integral extends over the entire feature space.

Minimizing $ R(\alpha(\mathbf{x}) | \mathbf{x}) $ for each $ \mathbf{x}  ensures that $ R $ is minimized, achieving optimal decision-making under uncertainty.


```python
# Bayesian Decision Theory Implementation

# Define the loss function
def loss(action, state):
    # Example: Simple loss function, can be customized
    return 1 if action != state else 0  # Cost of 1 for misclassification

# Function to compute the class-conditional probabilities
def class_conditional_probability(x, state):
    # Example: Assuming a Gaussian distribution for demonstration
    mean = 10 if state == 'ω1' else 12
    variance = 1
    from math import exp, sqrt, pi
    return (1 / sqrt(2 * pi * variance)) * exp(-((x - mean) ** 2) / (2 * variance))

# Function to compute prior probabilities
def prior_probability(state):
    # Example: Assuming priors P(ω1) = 2/3 and P(ω2) = 1/3
    return 2/3 if state == 'ω1' else 1/3

# Function to compute posterior probabilities using Bayes' theorem
def posterior_probability(x, state):
    numerator = class_conditional_probability(x, state) * prior_probability(state)
    evidence = sum(class_conditional_probability(x, s) * prior_probability(s) for s in ['ω1', 'ω2'])
    return numerator / evidence

# Function to compute the expected loss (conditional risk) for an action
def conditional_risk(action, x):
    expected_loss = 0
    for state in ['ω1', 'ω2']:
        expected_loss += loss(action, state) * posterior_probability(x, state)
    return expected_loss

# Decision function based on minimizing expected loss
def bayesian_decision(x):
    actions = ['ω1', 'ω2']  # Possible actions
    risks = {action: conditional_risk(action, x) for action in actions}
    return min(risks, key=risks.get)  # Return the action with minimum risk

# Example usage
if __name__ == "__main__":
    feature_value = 14  # Example feature value
    decision = bayesian_decision(feature_value)
    print(f"Decision for feature value {feature_value}: {decision}")
```

## 2.2.1 Two-Category Classification

Consider the application of decision rules to two-category classification problems. Here, action $\alpha_1$ corresponds to deciding that the true state of nature is $\omega_1$, and action $\alpha_2$ corresponds to deciding it is $\omega_2$. For simplicity, let $\lambda_{ij} = \lambda(\alpha_i | \omega_j)$ represent the loss incurred for deciding $\omega_i$ when the true state is $\omega_j$.

### Conditional Risk Derivation
The conditional risks can be expressed as:
$$
R(\alpha_1 | x) = \lambda_{11} P(\omega_1 | x) + \lambda_{12} P(\omega_2 | x)
$$
$$
R(\alpha_2 | x) = \lambda_{21} P(\omega_1 | x) + \lambda_{22} P(\omega_2 | x)
$$

The decision rule is to choose $\omega_1$ if $R(\alpha_1 | x) < R(\alpha_2 | x)$. This leads to the condition:
$$
(\lambda_{21} - \lambda_{11}) P(\omega_1 | x) > (\lambda_{12} - \lambda_{22}) P(\omega_2 | x)
$$

### Simplified Decision Rule
Assuming typical conditions where $\lambda_{21} > \lambda_{11}$ and $\lambda_{12} > \lambda_{22}$, we decide based on the more probable state scaled by the loss differences. Using Bayes’ theorem, we substitute the posterior probabilities to get:
$$
(\lambda_{21} - \lambda_{11}) p(x | \omega_1) P(\omega_1) > (\lambda_{12} - \lambda_{22}) p(x | \omega_2) P(\omega_2)
$$

### Likelihood Ratio Interpretation
Under the assumption $\lambda_{21} > \lambda_{11}$, we can rewrite this as:
$$
\frac{p(x | \omega_1)}{p(x | \omega_2)} > \frac{\lambda_{12} - \lambda_{22}}{\lambda_{21} - \lambda_{11}} \cdot \frac{P(\omega_2)}{P(\omega_1)}
$$

This indicates that the decision to choose $\omega_1$ depends on whether the likelihood ratio exceeds a threshold independent of $x$.

## 2.3 Minimum-Error-Rate Classification

In classification problems, each state $\omega_i$ is associated with a class, and action $\alpha_i$ corresponds to deciding the true state is $\omega_i$. When errors must be minimized, a **zero-one loss function** is used:
$$
\lambda(\alpha_i | \omega_j) = \begin{cases} 
0, & \text{if } i = j \\
1, & \text{if } i \neq j
\end{cases}
$$

### Conditional Risk with Zero-One Loss
The risk becomes:
$$
R(\alpha_i | x) = \sum_{j=1}^{c} \lambda(\alpha_i | \omega_j) P(\omega_j | x)
$$
$$
= 1 - P(\omega_i | x)
$$

### Bayes Decision Rule for Minimum Error
To minimize the error rate, choose $i$ such that $P(\omega_i | x)$ is maximized:
$$
\text{Decide } \omega_i \text{ if } P(\omega_i | x) > P(\omega_j | x) \text{ for all } j \neq i
$$

This decision rule aligns with minimizing conditional risk and results in decision boundaries consistent with those derived using likelihood ratios. Figure 2.3 illustrates the likelihood ratio $\frac{p(x | \omega_1)}{p(x | \omega_2)}$ with different thresholds $\theta_a$ and $\theta_b$, depending on the loss penalties for classification errors.




Fig.3: The likelihood ratio p(x|ω1 )/p(x|ω2 ) for the distributions shown in Fig.1. If we employ a zero-one or classiﬁcation loss, our decision boundaries are determined by the threshold θa . If our loss function penalizes miscategorizing ω2 as ω1 patterns more than the converse, (i.e., λ12 > λ21 ), we get the larger threshold θb , and hence R1 becomes smaller.

```python
# Define prior probabilities for classes ω1 and ω2
P_omega1 = 0.6  # Prior probability of ω1
P_omega2 = 0.4  # Prior probability of ω2

# Define likelihood functions p(x|ω1) and p(x|ω2)
def p_x_given_omega1(x):
    # Example: Gaussian distribution centered at 3 with variance 1
    from math import exp, sqrt, pi
    return (1 / sqrt(2 * pi)) * exp(-0.5 * (x - 3)**2)

def p_x_given_omega2(x):
    # Example: Gaussian distribution centered at 6 with variance 1
    from math import exp, sqrt, pi
    return (1 / sqrt(2 * pi)) * exp(-0.5 * (x - 6)**2)

# Define loss values λij
lambda_11 = 0  # No loss for correct decision of ω1
lambda_12 = 1  # Loss for deciding ω1 when true state is ω2
lambda_21 = 2  # Loss for deciding ω2 when true state is ω1
lambda_22 = 0  # No loss for correct decision of ω2

# Compute the decision rule threshold
threshold = (lambda_12 - lambda_22) / (lambda_21 - lambda_11) * (P_omega2 / P_omega1)

# Evaluate the decision rule for a given x
def decide_class(x):
    likelihood_ratio = p_x_given_omega1(x) / p_x_given_omega2(x)
    if likelihood_ratio > threshold:
        return 'Decide ω1'
    else:
        return 'Decide ω2'

# Test the decision function for a range of x values
x_values = [1, 3, 5, 7, 9]
for x in x_values:
    decision = decide_class(x)
    print(f"x = {x}: {decision}")
```

##  Minimax Criterion

Sometimes we must design our classifier to perform well over a range of prior probabilities. For instance, in our fish categorization problem, we can imagine that whereas the physical properties of lightness and width of each type of fish remain constant, the prior probabilities might vary widely and in an unpredictable way, or alternatively, we want to use the classifier in a different plant where we do not know the prior probabilities.

A reasonable approach is then to design our classifier so that the worst overall risk for any value of the priors is as small as possible — that is, minimize the maximum possible overall risk.

To understand this, we let $ R_1 $ denote the region in feature space where the classifier decides $ \omega_1 $ and similarly for $ R_2 $ and $ \omega_2 $. We write the overall risk $ R $ in terms of conditional risks:

$$
R = \int_{R_1} [\lambda_{11} P(\omega_1) p(x|\omega_1) + \lambda_{12} P(\omega_2) p(x|\omega_2)] \, dx + \int_{R_2} [\lambda_{21} P(\omega_1) p(x|\omega_1) + \lambda_{22} P(\omega_2) p(x|\omega_2)] \, dx.
$$

Using the fact that $ P(\omega_2) = 1 - P(\omega_1) $ and $ \int_{R_1} p(x|\omega_1) \, dx = 1 - \int_{R_2} p(x|\omega_1) \, dx $, we rewrite the risk as:

$$
R(P(\omega_1)) = \lambda_{22} + (\lambda_{12} - \lambda_{22}) \int_{R_1} p(x|\omega_2) \, dx + P(\omega_1) \left[(\lambda_{11} - \lambda_{22}) - (\lambda_{21} - \lambda_{11}) \int_{R_2} p(x|\omega_1) \, dx - (\lambda_{12} - \lambda_{22}) \int_{R_1} p(x|\omega_2) \, dx\right].
$$

For a minimax solution, we aim for:

$$
\text{minimax risk} \quad R_{mm} = \lambda_{22} + (\lambda_{12} - \lambda_{22}) \int_{R_1} p(x|\omega_2) \, dx = \lambda_{11} + (\lambda_{21} - \lambda_{11}) \int_{R_2} p(x|\omega_1) \, dx.
$$

For any given boundary, if the priors change, the probability of error changes linearly (dashed line). To minimize the maximum error, the decision boundary should be designed for the maximum Bayes error, ensuring the risk is constant (solid line).


```python
import numpy as np
import matplotlib.pyplot as plt

# Define synthetic data for class-conditional distributions
def p_x_given_w1(x):
    # Assume a Gaussian distribution for ω1
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * (x - 1)**2)

def p_x_given_w2(x):
    # Assume a Gaussian distribution for ω2
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * (x + 1)**2)

# Define prior probabilities (changeable to simulate minimax criterion)
P_w1 = 0.5
P_w2 = 1 - P_w1

# Define loss matrix elements
lambda_11, lambda_12 = 0, 1  # Cost of correct decision and misclassification
lambda_21, lambda_22 = 1, 0

# Define likelihood ratios and risk calculations
x_values = np.linspace(-4, 4, 1000)
risk_decide_w1 = lambda_11 * P_w1 * p_x_given_w1(x_values) + lambda_12 * P_w2 * p_x_given_w2(x_values)
risk_decide_w2 = lambda_21 * P_w1 * p_x_given_w1(x_values) + lambda_22 * P_w2 * p_x_given_w2(x_values)

# Plot risk functions
plt.figure(figsize=(10, 6))
plt.plot(x_values, risk_decide_w1, label='Risk for decision ω1')
plt.plot(x_values, risk_decide_w2, label='Risk for decision ω2')
plt.axhline(y=np.min([np.max(risk_decide_w1), np.max(risk_decide_w2)]), color='r', linestyle='--', label='Minimax risk threshold')
plt.xlabel('x')
plt.ylabel('Risk')
plt.title('Minimax Risk Evaluation for Two-Category Classification')
plt.legend()
plt.grid(True)
plt.show()
```



Fig.4: The curve at the bottom shows the minimum (Bayes) error as a function of prior probability P (ω1 ) in a two-category classiﬁcation problem of ﬁxed distributions. For each value of the priors (e.g., P (ω1 ) = 0.25) there is a corresponding optimal decision boundary and associated Bayes error rate. For any (ﬁxed) such boundary, if the priors are then changed, the probability of error will change as a linear function of P (ω1 ) (shown by the dashed line). The maximum such error will occur at an extreme value of the prior, here at P (ω1 ) = 1. To minimize the maximum of such error, we should design our decision boundary for the maximum Bayes error (here P (ω1 ) = 0.6), and thus the error will not change as a function of prior, as shown by the solid red horizontal line.

## Neyman-Pearson Criterion and Classifier Implementation

## Neyman-Pearson Criterion

In some problems, we may wish to minimize the overall risk subject to a constraint. For instance, we might seek to minimize the total risk while ensuring:

$$
\int_{R(\alpha_i | x)} R(\alpha_i | x) \, dx < \text{constant}
$$

for some particular $ i $. This constraint may arise if a fixed resource accompanies a particular action $ \alpha_i $, or if misclassification from a certain state of nature $ \omega_i $ must not exceed a specified frequency. For example, suppose government regulations stipulate that no more than 1% of salmon should be misclassified as sea bass. The decision strategy would then aim to minimize the risk of classifying a sea bass as salmon under this condition.

Neyman-Pearson criteria often require numerical adjustment of decision boundaries. However, analytical solutions may be available for Gaussian or other specific distributions.

## 2.4 Classifiers, Discriminant Functions, and Decision Surfaces

### 2.4.1 The Multi-Category Case

Pattern classifiers can be represented by a set of discriminant functions $ g_i(x) $ for $ i = 1, \ldots, c $. The classifier assigns a feature vector $ x $ to class $ \omega_i $ if:

$$
g_i(x) > g_j(x) \quad \text{for all } j \neq i.
$$

This concept is visualized in the structure of a statistical pattern classifier network, as shown below:



Fig.5: The functional structure of a general statistical pattern classiﬁer which includes d inputs and c discriminant functions gi (x). A subsequent step determines which of the discriminant values is the maximum, and categorizes the input pat- tern accordingly. The arrows show the direction of the ﬂow of information, though frequently the arrows are omitted when the direction of ﬂow is self-evident.

### Example Representation:

A classifier receives input $ x = [x_1, x_2, \ldots, x_d] $, computes discriminant functions, and assigns the class corresponding to the maximum discriminant.

### Bayes Classifier Representation:

A Bayes classifier can be represented with:

$$
g_i(x) = -R(\alpha_i | x)
$$

where the maximum discriminant function corresponds to the minimum conditional risk. For minimum-error-rate classification:

$$
g_i(x) = P(\omega_i | x)
$$

so the decision rule selects the maximum posterior probability. The choice of discriminant functions can be modified without affecting classification by:

$$
g_i(x) = f(g_i(x))
$$

where $ f(\cdot) $ is a monotonically increasing function. This leads to the following equivalent forms:

$$
g_i(x) = \frac{p(x|\omega_i)P(\omega_i)}{\sum_{j=1}^{c} p(x|\omega_j)P(\omega_j)}
$$

or

$$
g_i(x) = p(x|\omega_i)P(\omega_i)
$$

or

$$
g_i(x) = \ln p(x|\omega_i) + \ln P(\omega_i).
$$

These forms simplify analysis and computation while maintaining equivalent classification results.

### Decision Regions and Boundaries:

The decision rules divide the feature space into regions $ R_1, R_2, \ldots, R_c $. A feature vector $ x $ belongs to region $ R_i $ if:

$$
g_i(x) > g_j(x) \quad \text{for all } j \neq i.
$$

Decision boundaries are the surfaces where discriminant functions tie.

### Visualization of Decision Boundaries:

For Gaussian distributions, decision boundaries may be hyperbolic, leading to non-simply connected decision regions.



Fig.6: In this two-dimensional two-category classiﬁer, the probability densities are Gaussian (with 1/e ellipses shown), the decision boundary consists of two hyper- bolas, and thus the decision region R2 is not simply connected.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic data for demonstration
np.random.seed(42)

# Class means and covariances
means = [(-2, -2), (2, 2), (-2, 2), (2, -2)]
covariances = [np.array([[1, 0.5], [0.5, 1]]), np.array([[1, -0.5], [-0.5, 1]]),
               np.array([[1, 0], [0, 1]]), np.array([[1, 0.3], [0.3, 1]])]

# Number of samples per class
num_samples = 100

# Generate data
data = []
labels = []
for i, (mean, cov) in enumerate(zip(means, covariances)):
    samples = np.random.multivariate_normal(mean, cov, num_samples)
    data.append(samples)
    labels.extend([i] * num_samples)

data = np.vstack(data)
labels = np.array(labels)

# Create a grid for decision boundary visualization
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

# Compute Gaussian probability density function manually
def gaussian_pdf(x, mean, cov):
    d = mean.shape[0]  # Number of dimensions
    cov_inv = np.linalg.inv(cov)  # Inverse of covariance
    det_cov = np.linalg.det(cov)  # Determinant of covariance
    norm_const = 1 / ((2 * np.pi) ** (d / 2) * (det_cov ** 0.5))  # Normalization constant
    x_mu = x - mean  # Deviation from mean
    result = norm_const * np.exp(-0.5 * np.dot(np.dot(x_mu, cov_inv), x_mu.T))
    return result

# Compute discriminant functions for each class
Z = np.zeros((xx.shape[0], xx.shape[1], len(means)))
for i, (mean, cov) in enumerate(zip(means, covariances)):
    for j in range(xx.shape[0]):
        for k in range(xx.shape[1]):
            Z[j, k, i] = gaussian_pdf(np.array([xx[j, k], yy[j, k]]), np.array(mean), cov)

# Assign each point to the class with the maximum discriminant function
Z_max = np.argmax(Z, axis=2)

# Prepare the decision boundary data for Seaborn
boundary_data = np.c_[xx.ravel(), yy.ravel(), Z_max.ravel()]
boundary_df = pd.DataFrame(boundary_data, columns=['x', 'y', 'class'])

# Plot using Seaborn
plt.figure(figsize=(10, 8))
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, palette='deep', alpha=0.6)
sns.scatterplot(data=boundary_df, x='x', y='y', hue='class', palette='deep', alpha=0.1, legend=False)

plt.title("Multi-Class Classifier with Gaussian Distributions")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title='Class')
plt.show()
```
