## Iterative Optimization for Clustering

Once a criterion function has been selected, clustering becomes a well-defined problem in discrete optimization: find those partitions of the set of samples that extremize the criterion function. Since the sample set is finite, there are only a finite number of possible partitions. Thus, in theory, the clustering problem can always be solved by exhaustive enumeration. However, the computational complexity renders such an approach unthinkable for all but the simplest problems; there are approximately:

$$ \frac{c^n}{c!} $$

ways of partitioning a set of $n$ elements into $c$ subsets, and this exponential growth with $n$ is overwhelming (Problem 17). For example, an exhaustive search for the best set of 5 clusters in 100 samples would require considering more than $10^{67}$ partitionings. Simply put, in most applications an exhaustive search is completely infeasible.

The approach most frequently used in seeking optimal partitions is iterative optimization. The basic idea is to find some reasonable initial partition and to "move" samples from one group to another if such a move will improve the value of the criterion function. Like hill-climbing procedures in general, these approaches guarantee local but not global optimization. Different starting points can lead to different solutions, and one never knows whether or not the best solution has been found. Despite these limitations, the fact that the computational requirements are bearable makes this approach attractive.

Let us consider the use of iterative improvement to minimize the sum-of-squared-error criterion $ J_e $, written as:

$$ J_e = \sum_{i=1}^{c} J_i $$

where an effective error per cluster is defined to be:

$$ J_i = \sum_{x \in D_i} (x - m_i)^2 $$

and the mean of each cluster is, as before:

$$ m_i = \frac{1}{n_i} \sum_{x \in D_i} x $$

Suppose that a sample $ \hat{x} $ currently in cluster $ D_i $ is tentatively moved to $ D_j $. Then $ m_j $ changes to:

$$ m_j^* = m_j + \frac{x - m_j}{n_j + 1} $$

and $ J_j $ increases to:

$$ J_j^* = \sum_{x \in D_j} (x - m_j^*)^2 + (x - m_j)^2 $$

Under the assumption that $ n_i \neq 1 $ (singleton clusters should not be destroyed), a similar calculation shows that $ m_i $ changes to:

$$ m_i^* = m_i - \frac{x - m_i}{n_i - 1} $$

and \( J_i \) decreases to:

$$ J_i^* = J_i - (x - m_i)^2 $$

These equations greatly simplify the computation of the change in the criterion function. The transfer of $ \hat{x} $ from $ D_i $ to $ D_j $ is advantageous if the decrease in $ J_i $ is greater than the increase in $ J_j $. This is the case if:

$$ \frac{n_i}{n_i - 1} (x - m_i)^2 > \frac{n_j}{n_j + 1} (x - m_j)^2 $$

which typically happens whenever $ \hat{x} $ is closer to $ m_j $ than $ m_i $. If reassignment is profitable, the greatest decrease in the sum of squared error is obtained by selecting the cluster for which:

$$ \frac{n_j}{n_j + 1} (x - m_j)^2 $$

is minimum. This leads to the following clustering procedure:

### Algorithm 3 (Basic Iterative Minimum-Squared-Error Clustering)

1. **Initialize** $ n, c, m_1, m_2, \dots, m_c $
2. **Repeat**:
    - Randomly select a sample $ \hat{x} $
    - Classify $ \hat{x} $ by finding the cluster $ i $ that minimizes $ \| m_i - \hat{x} \| $
    - If $ n_i \neq 1 $, compute the transfer cost for each cluster
    - Transfer $ \hat{x} $ to the cluster with the lowest transfer cost
    - Recompute $ J_e $ and the cluster means
3. **Until** no change in $ J_e $ for $ n $ attempts
4. **Return** the final cluster means $ m_1, m_2, \dots, m_c $

A moment’s consideration will show that this procedure is essentially a sequential version of the k-means procedure. Where the k-means procedure waits until all $ n $ samples have been reclassified before updating, the Basic Iterative Minimum-Squared-Error procedure updates after each sample is reclassified.

This procedure has been experimentally observed to be more susceptible to being trapped in local minima, and it has the further disadvantage of making the results depend on the order in which the candidates are selected. However, it is at least a stepwise optimal procedure, and it can be easily modified to apply to problems in which samples are acquired sequentially and clustering must be done on-line.

### Choosing the Starting Point

One question that plagues all hill-climbing procedures is the choice of the starting point. Unfortunately, there is no simple, universally good solution to this problem. One approach is to select $ c $ samples randomly for the initial cluster centers, using them to partition the data on a minimum-distance basis. Repetition with different random selections can give some indication of the sensitivity of the solution to the starting point.

Another approach is to find the $ c $-cluster starting point from the solutions to the $ (c - a) $-cluster problem. The solution for the one-cluster problem is the total sample mean; the starting point for the $ c $-cluster problem can be the final means for the $ (c - a) $-cluster problem plus the sample that is farthest from the nearest cluster center. This approach leads us directly to the so-called hierarchical clustering procedures, which are simple methods that can provide very good starting points for iterative optimization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a synthetic dataset
np.random.seed(42)
n_samples = 300
X = np.vstack([
    np.random.randn(n_samples // 3, 2) + np.array([5, 5]),
    np.random.randn(n_samples // 3, 2) + np.array([0, -5]),
    np.random.randn(n_samples // 3, 2) + np.array([-5, 5])
])

# Step 2: Implement the Basic Iterative Minimum-Squared-Error Clustering
def basic_iterative_min_squared_error_clustering(X, c, max_iterations=100, tolerance=1e-6):
    n_samples, n_features = X.shape
    
    # Step 1: Randomly initialize cluster centers (means)
    indices = np.random.choice(n_samples, c, replace=False)
    m = X[indices]  # Initialize cluster centers
    
    # Function to compute the sum of squared errors (criterion)
    def compute_criterion(cluster_assignments):
        J = 0
        for i in range(c):
            # For each cluster, calculate the sum of squared errors
            cluster_points = X[cluster_assignments == i]
            J += np.sum(np.linalg.norm(cluster_points - m[i], axis=1) ** 2)
        return J
    
    criterion = compute_criterion(np.zeros(n_samples))  # Initial assignment (dummy)
    
    # Step 3: Iterate to optimize the criterion function
    history = []  # To store the cluster centers for visualization
    for iteration in range(max_iterations):
        criterion_old = criterion
        
        # Step 2: Assign each point to the closest cluster center
        distances = np.linalg.norm(X[:, np.newaxis] - m, axis=2)  # Shape: (n_samples, c)
        cluster_assignments = np.argmin(distances, axis=1)
        
        # Step 3: Recompute the mean of each cluster
        for i in range(c):
            m[i] = np.mean(X[cluster_assignments == i], axis=0)
        
        # Compute new criterion function value
        criterion = compute_criterion(cluster_assignments)
        
        # Record the cluster centers for visualization after this iteration
        history.append(m.copy())
        
        # Stop if the change in criterion function is very small
        if np.abs(criterion - criterion_old) < tolerance:
            break
    
    return m, history, cluster_assignments

# Step 3: Run the algorithm with 3 clusters
c = 3
final_centers, history, cluster_assignments = basic_iterative_min_squared_error_clustering(X, c)

# Step 4: Visualize the results
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# Plot the data points with colors based on cluster assignments
scatter = ax.scatter(X[:, 0], X[:, 1], c=cluster_assignments, s=30, alpha=0.5, cmap='viridis', label='Data points')

# Plot the cluster centers over iterations
for idx, centers in enumerate(history):
    ax.scatter(centers[:, 0], centers[:, 1], marker='x', label=f'Iteration {idx+1}', s=100)

# Final cluster centers
ax.scatter(final_centers[:, 0], final_centers[:, 1], color='red', marker='x', s=200, label='Final centers')

ax.set_title("Basic Iterative Minimum-Squared-Error Clustering")
ax.legend()
plt.show()

# Print final cluster centers
print("Final Cluster Centers:")
print(final_centers)
```



##  Hierarchical Clustering

Up to now, our methods have formed disjoint clusters — in computer science terminology, we would say that the data description is “flat.” However, there are many times when clusters have subclusters, these have sub-subclusters, and so on. In biological taxonomy, for instance, kingdoms are split into phylums, which are split into subphyla, which are split into orders, and suborders, and families, and subfamilies, and genus and species, and so on, all the way to a particular individual organism. Thus we might have kingdom = animal, phylum = Chordata, subphylum = Vertebrata, class = Osteichthyes, subclass = Actinopterygii, order = Salmoniformes, family = Salmonidae, genus = Oncorhynchus, species = Oncorhynchus kisutch, and individual = the particular Coho salmon caught in my net.

Organisms that lie in the animal kingdom — such as a salmon and a moose — share important attributes that are not present in organisms in the plant kingdom, such as redwood trees. In fact, this kind of hierarchical clustering permeates classifying activities in the sciences. Thus, we now turn to clustering methods which will lead to representations that are “hierarchical,” rather than flat.

## Definitions

Let us consider a sequence of partitions of the $ n $ samples into $ c $ clusters. The first of these is a partition into $ n $ clusters, each cluster containing exactly one sample. The next is a partition into $ n - 1 $ clusters, the next a partition into $ n - 2 $, and so on until the $ n $-th, in which all the samples form one cluster. We shall say that we are at level $ k $ in the sequence when $ c = n - k + 1 $. Thus, level one corresponds to $ n $ clusters, and level $ n $ to one cluster.

Given any two samples $ x $ and $ x' $, at some level they will be grouped together in the same cluster. If the sequence has the property that whenever two samples are in the same cluster at level $ k $, they remain together at all higher levels, then the sequence is said to be a hierarchical clustering. The most natural representation of hierarchical clustering is a corresponding tree, called a dendrogram, which shows how the samples are grouped.

### Dendrogram Representation

A dendrogram is a tree-like diagram that represents the hierarchical relationship between clusters. The vertical axis of the dendrogram typically shows the level of similarity between clusters, with higher positions indicating more similarity between clusters that are merged. The levels of the dendrogram represent the progressive merging of clusters, starting with each sample in its own cluster at the top and progressively merging clusters as you move down.

Consider the following dendrogram showing how samples are clustered at different levels of similarity:

$$
\begin{array}{c|c}
\text{Level} & \text{Similarity} \\
\hline
1 & \text{All samples as singleton clusters (highest similarity)} \\
2 & \text{First pair of samples merged into a cluster} \\
3 & \text{Another pair merged, and so on} \\
4 & \text{Clusters continue merging as the similarity decreases} \\
5 & \text{Final cluster (lowest similarity)} \\
\end{array}
$$

### Example Dendrogram

The dendrogram in Figure 10.10 illustrates how clusters are merged. At level 1, all eight samples are in singleton clusters. At level 2, samples $ x_6 $ and $ x_7 $ are merged, and this clustering persists at all higher levels.

### Set or Venn Diagram Representation

Another way to represent hierarchical clustering is using sets or Venn diagrams. In these diagrams, each cluster can contain subclusters, revealing the hierarchical structure. However, this representation does not clearly indicate the quantitative similarity between clusters, which is why dendrograms are generally preferred for visualizing hierarchical clustering.

### Types of Hierarchical Clustering

There are two main approaches to hierarchical clustering:

- **Agglomerative (Bottom-Up)**: 
  - Start with each sample as its own cluster.
  - Merge the two most similar clusters at each step.
  - This approach is simpler computationally for small datasets but can become computationally expensive for large datasets.

- **Divisive (Top-Down)**: 
  - Start with all samples in one cluster.
  - Recursively split the most dissimilar cluster into smaller clusters.
  - This approach is less commonly used because of the higher computational complexity.

### Visualizing the Hierarchical Clustering Process

The agglomerative approach is the most common and will be illustrated with a simple example. Here, we start with each sample in its own cluster, and at each step, we merge the two clusters that are most similar to each other, based on a predefined similarity measure.

At each level, we update the dendrogram to reflect the merging of clusters.

## Conclusion

Hierarchical clustering provides a way to organize data in a hierarchy, which is useful in situations where natural subclusters exist. Dendrograms are an effective tool to visualize hierarchical clustering, and the agglomerative method is widely used due to its simplicity. By examining the dendrogram, we can gain insights into the natural groupings within the data and determine an appropriate number of clusters based on the similarity values.

##  Agglomerative Hierarchical Clustering

The major steps in agglomerative clustering are contained in the following procedure, where $ c $ is the desired number of final clusters:

## Algorithm : Agglomerative Hierarchical Clustering

1. **Initialize:**
   - $ c, \hat{c} \leftarrow n $
   - $ D_i \leftarrow \{x_i\} $, for $ i = 1, \dots, n $

2. **Repeat:**
   - $ \hat{c} \leftarrow \hat{c} - 1 $
   - Find nearest clusters, say $ D_i $ and $ D_j $
   - Merge $ D_i $ and $ D_j $

3. **Terminate** when $ c = \hat{c} $

4. **Return:** $ c $ clusters

This procedure terminates when the specified number of clusters has been obtained and returns the clusters, described as sets of points (rather than as mean or representative vectors). If we continue until $ c = 1 $, we can produce a dendrogram like the one in Fig.10.

### Distance Measures for Cluster Merger

At any level, the "distance" between the nearest clusters can provide the dissimilarity value for that level. The distance measures used to compute the distance between two clusters are as follows:

- **Minimum Distance** $ d_{\text{min}}(D_i, D_j) $:

$$
d_{\text{min}}(D_i, D_j) = \min_{x \in D_i, x' \in D_j} \|x - x'\|
$$

- **Maximum Distance** $ d_{\text{max}}(D_i, D_j) $:

$$
d_{\text{max}}(D_i, D_j) = \max_{x \in D_i, x' \in D_j} \|x - x'\|
$$

- **Average Distance** $ d_{\text{avg}}(D_i, D_j) $:

$$
d_{\text{avg}}(D_i, D_j) = \frac{1}{n_i n_j} \sum_{x \in D_i, x' \in D_j} \|x - x'\|
$$

- **Mean Distance** \( d_{\text{mean}}(D_i, D_j) \):

$$
d_{\text{mean}}(D_i, D_j) = \| m_i - m_j \|
$$

where $ m_i $ and $ m_j $ represent the centroids (mean vectors) of clusters $ D_i $ and $ D_j $, respectively.

These measures generally yield the same results when clusters are compact and well separated. However, when clusters are close or not hyperspherical in shape, they can give different results.

### Computational Complexity

Consider a scenario where we have $ n $ patterns in a $ d $-dimensional space and seek to form $ c $ clusters using the $ d_{\text{min}} $ distance measure. The complexity involved is as follows:

- **Space complexity**: $ O(n^2) $ — We need to calculate $ n(n - 1) $ inter-point distances, each of which requires an $ O(d^2) $ operation.
- **Time complexity for the first agglomerative step**: $ O(n^2 d^2) $ — This is due to the need to search through the distance table for the minimum distance pair.
- **Time complexity for subsequent steps**: $ O(n(n - 1) - \hat{c}) $, where we step through the "unused" distances to find the nearest clusters.
- **Total time complexity**: $ O(cn^2 d^2) $, where typically $ n \gg c 4.

### The Nearest-Neighbor Algorithm

When $ d_{\text{min}} $ is used to measure the distance between clusters, the algorithm is known as the **nearest-neighbor clustering algorithm** or **minimum algorithm**. If terminated when the distance between nearest clusters exceeds a threshold, it is referred to as the **single-linkage algorithm**.

We can think of the data points as nodes of a graph, with edges forming a path between nodes within the same subset $ D_i $. When $ d_{\text{min}} $ is used, the nearest nodes in different clusters determine the nearest clusters. The merging of two clusters corresponds to adding an edge between the closest pair of nodes in the two clusters. This process produces a tree, known as a **spanning tree**. The sum of edge lengths in this tree will not exceed the sum of edges in any other spanning tree.

### The Farthest-Neighbor Algorithm

When $ d_{\text{max}} $ is used to measure the distance between subsets, the algorithm is called the **farthest-neighbor clustering algorithm** or **maximum algorithm**. When it is terminated based on an arbitrary threshold between nearest clusters, it is referred to as the **complete-linkage algorithm**.

The farthest-neighbor algorithm discourages the growth of elongated clusters. Every cluster is considered a complete subgraph in graph theory. The distance between two clusters is determined by the most distant nodes in the two clusters.

The merging of two clusters involves adding edges between all pairs of nodes in the two clusters. The goal is to minimize the increase in the diameter of the partition. This is advantageous when the true clusters are compact and roughly equal in size. However, when clusters are elongated, as illustrated in Fig.13, the resulting groupings can be less meaningful.

### Graphical Example

The figures below show how the agglomerative clustering procedure works using Gaussian data.

This behavior, called the "chaining effect," can result in poor clustering when the data contains noise or is not well separated.

### Conclusion

Agglomerative hierarchical clustering is a bottom-up approach where clusters are successively merged based on a distance measure. The nearest-neighbor and farthest-neighbor algorithms are two key variants, each having its strengths and weaknesses depending on the shape and distribution of the data. In practice, agglomerative clustering can be effective for well-separated, compact clusters, but may struggle with elongated or noisy data.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from scipy.cluster.hierarchy import dendrogram, linkage

# Function to compute the minimum distance (nearest-neighbor)
def dmin(Di, Dj):
    return np.min(cdist(Di, Dj))

# Function to compute the maximum distance (farthest-neighbor)
def dmax(Di, Dj):
    return np.max(cdist(Di, Dj))

# Function to compute the average distance between two clusters
def davg(Di, Dj):
    return np.mean(cdist(Di, Dj))

# Function to compute the mean distance between two clusters
def dmean(Di, Dj):
    return np.linalg.norm(np.mean(Di, axis=0) - np.mean(Dj, axis=0))

# Agglomerative Hierarchical Clustering Implementation
def agglomerative_clustering(X, c, distance_metric='min'):
    # Initialize clusters as singleton clusters
    clusters = [[i] for i in range(X.shape[0])]
    distances = cdist(X, X)  # Initial distance matrix
    
    # Define distance metric for merging clusters
    if distance_metric == 'min':
        distance_func = dmin
    elif distance_metric == 'max':
        distance_func = dmax
    elif distance_metric == 'avg':
        distance_func = davg
    elif distance_metric == 'mean':
        distance_func = dmean
    else:
        raise ValueError("Invalid distance metric selected.")
    
    # Hierarchical clustering loop
    while len(clusters) > c:
        min_dist = float('inf')
        merge_pair = (0, 0)
        
        # Find the nearest clusters
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                dist = distance_func(X[clusters[i]], X[clusters[j]])
                if dist < min_dist:
                    min_dist = dist
                    merge_pair = (i, j)
        
        # Merge the clusters
        new_cluster = clusters[merge_pair[0]] + clusters[merge_pair[1]]
        clusters = [clusters[k] for k in range(len(clusters)) if k != merge_pair[0] and k != merge_pair[1]]
        clusters.append(new_cluster)
        
    # Return final clusters
    return clusters

# Example usage
from sklearn.datasets import make_blobs

# Create a sample dataset
X, y = make_blobs(n_samples=10, centers=3, random_state=42)

# Perform agglomerative hierarchical clustering
c = 3  # Desired number of clusters
clusters = agglomerative_clustering(X, c, distance_metric='min')

# Visualize the final clustering
colors = ['red', 'green', 'blue', 'purple', 'orange']
plt.figure(figsize=(8, 6))

for i, cluster in enumerate(clusters):
    cluster_points = X[cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f'Cluster {i+1}')

plt.title('Agglomerative Hierarchical Clustering')
plt.legend()
plt.show()

# Generate a dendrogram to visualize the clustering process
linked = linkage(X, method='single')
plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title('Dendrogram of Agglomerative Clustering')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute distance functions between clusters

# Minimum distance (Nearest Neighbor)
def dmin(Di, Dj, X):
    # Calculate minimum distance between two clusters Di and Dj
    min_dist = np.inf
    for i in Di:
        for j in Dj:
            dist = np.linalg.norm(X[i] - X[j])
            min_dist = min(min_dist, dist)
    return min_dist

# Maximum distance (Farthest Neighbor)
def dmax(Di, Dj, X):
    # Calculate maximum distance between two clusters Di and Dj
    max_dist = -np.inf
    for i in Di:
        for j in Dj:
            dist = np.linalg.norm(X[i] - X[j])
            max_dist = max(max_dist, dist)
    return max_dist

# Average distance
def davg(Di, Dj, X):
    # Calculate average distance between two clusters Di and Dj
    dist = []
    for i in Di:
        for j in Dj:
            dist.append(np.linalg.norm(X[i] - X[j]))
    return np.mean(dist)

# Mean distance
def dmean(Di, Dj, X):
    # Calculate the mean of centroids' distance between two clusters Di and Dj
    centroid_i = np.mean(X[Di], axis=0)
    centroid_j = np.mean(X[Dj], axis=0)
    return np.linalg.norm(centroid_i - centroid_j)

# Main Agglomerative Clustering Algorithm
def agglomerative_clustering(X, c, distance_metric='min'):
    n = X.shape[0]  # Number of samples
    clusters = [[i] for i in range(n)]  # Start with singleton clusters
    distance_matrix = np.zeros((n, n))  # Initialize distance matrix
    
    # Compute the initial distance matrix
    for i in range(n):
        for j in range(i + 1, n):
            distance_matrix[i, j] = np.linalg.norm(X[i] - X[j])
            distance_matrix[j, i] = distance_matrix[i, j]

    # Choose distance function based on the metric
    if distance_metric == 'min':
        distance_func = dmin
    elif distance_metric == 'max':
        distance_func = dmax
    elif distance_metric == 'avg':
        distance_func = davg
    elif distance_metric == 'mean':
        distance_func = dmean
    else:
        raise ValueError("Invalid distance metric selected.")
    
    # Merge clusters until the desired number of clusters is reached
    while len(clusters) > c:
        min_dist = np.inf
        merge_pair = (0, 0)
        
        # Find the two closest clusters
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                dist = distance_func(clusters[i], clusters[j], X)
                if dist < min_dist:
                    min_dist = dist
                    merge_pair = (i, j)
        
        # Merge the closest pair of clusters
        new_cluster = clusters[merge_pair[0]] + clusters[merge_pair[1]]
        clusters = [clusters[k] for k in range(len(clusters)) if k != merge_pair[0] and k != merge_pair[1]]
        clusters.append(new_cluster)
    
    return clusters

# Visualize the final clustering
def visualize_clusters(X, clusters):
    plt.figure(figsize=(8, 6))
    colors = ['red', 'green', 'blue', 'purple', 'orange']
    
    for i, cluster in enumerate(clusters):
        cluster_points = X[cluster]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i % len(colors)], label=f'Cluster {i+1}')
    
    plt.title('Agglomerative Hierarchical Clustering')
    plt.legend()
    plt.show()

# Example usage: Create a synthetic dataset
from sklearn.datasets import make_blobs

# Create a sample dataset with 10 points and 3 centers
X, y = make_blobs(n_samples=10, centers=3, random_state=42)

# Perform agglomerative hierarchical clustering
c = 3  # Desired number of clusters
clusters = agglomerative_clustering(X, c, distance_metric='min')

# Visualize the final clustering
visualize_clusters(X, clusters)
```

```python
import matplotlib.pyplot as plt
import math

# Euclidean distance function
def euclidean_distance(point1, point2):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(point1, point2)))

# Agglomerative Hierarchical Clustering Algorithm
def agglomerative_clustering(X, c, distance_metric='min'):
    n = len(X)  # Number of samples
    clusters = [[i] for i in range(n)]  # Start with singleton clusters
    
    # Function to calculate distances between clusters based on the metric
    def calculate_distance(Di, Dj):
        if distance_metric == 'min':  # Minimum distance (Nearest Neighbor)
            return min(euclidean_distance(X[i], X[j]) for i in Di for j in Dj)
        elif distance_metric == 'max':  # Maximum distance (Farthest Neighbor)
            return max(euclidean_distance(X[i], X[j]) for i in Di for j in Dj)
        elif distance_metric == 'avg':  # Average distance
            dist = [euclidean_distance(X[i], X[j]) for i in Di for j in Dj]
            return sum(dist) / len(dist)
        elif distance_metric == 'mean':  # Mean distance (centroid)
            centroid_i = [sum(X[i][dim] for i in Di) / len(Di) for dim in range(len(X[0]))]
            centroid_j = [sum(X[i][dim] for i in Dj) / len(Dj) for dim in range(len(X[0]))]
            return euclidean_distance(centroid_i, centroid_j)
        else:
            raise ValueError("Invalid distance metric selected.")

    # Merge clusters until the desired number of clusters is reached
    while len(clusters) > c:
        min_dist = float('inf')
        merge_pair = (0, 0)
        
        # Find the two closest clusters
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                dist = calculate_distance(clusters[i], clusters[j])
                if dist < min_dist:
                    min_dist = dist
                    merge_pair = (i, j)
        
        # Merge the closest pair of clusters
        new_cluster = clusters[merge_pair[0]] + clusters[merge_pair[1]]
        clusters = [clusters[k] for k in range(len(clusters)) if k != merge_pair[0] and k != merge_pair[1]]
        clusters.append(new_cluster)

    return clusters

# Visualize the final clustering
def visualize_clusters(X, clusters):
    plt.figure(figsize=(8, 6))
    colors = ['red', 'green', 'blue', 'purple', 'orange']
    
    for i, cluster in enumerate(clusters):
        cluster_points = [X[i] for i in cluster]
        x_vals, y_vals = zip(*cluster_points)
        plt.scatter(x_vals, y_vals, color=colors[i % len(colors)], label=f'Cluster {i+1}')
    
    plt.title('Agglomerative Hierarchical Clustering')
    plt.legend()
    plt.show()

# Example usage: Create a synthetic dataset
from sklearn.datasets import make_blobs

# Create a sample dataset with 10 points and 3 centers
X, y = make_blobs(n_samples=10, centers=3, random_state=42)

# Perform agglomerative hierarchical clustering
c = 3  # Desired number of clusters
clusters = agglomerative_clustering(X, c, distance_metric='min')

# Visualize the final clustering
visualize_clusters(X, clusters)
```

## Compromises in Agglomerative Hierarchical Clustering

The **minimum** and **maximum** measures represent two extremes in measuring the distance between clusters. Like all procedures that involve minima or maxima, they tend to be overly sensitive to "outliers" or "wildshots". The use of averaging is an obvious way to ameliorate these problems, and $d_{\text{avg}}$ and $d_{\text{mean}}$ (Equations 76 & 77) are natural compromises between $d_{\text{min}}$ and $d_{\text{max}}$.

Computationally, $d_{\text{mean}}$ is the simplest of all of these measures, since the others require computing all $n_i n_j$ pairs of distances $x - x'$. However, a measure such as $d_{\text{avg}}$ can be used when the distances $x - x'$ are replaced by similarity measures, where the similarity between mean vectors may be difficult or impossible to define.

### Stepwise-Optimal Hierarchical Clustering

Earlier, we observed that if clusters are grown by merging the nearest pair of clusters, the results have a minimum variance flavor. However, when the measure $d_{\text{max}}$ is chosen arbitrarily, one can rarely assert that the resulting partition extremizes any particular criterion function.

In effect, hierarchical clustering defines a cluster as whatever results from applying the clustering procedure. Nevertheless, with a simple modification, it is possible to obtain a **stepwise-optimal procedure** for extremizing a criterion function. This is done by replacing line 3 of the Basic Iterative Agglomerative Clustering Procedure (Algorithm 4) by a more general form:

### Algorithm 5: Stepwise-Optimal Hierarchical Clustering
$$
\begin{array}{ll}
1. & \text{begin} \\
2. & \text{initialize } c, \hat{c} \leftarrow n, D_i \leftarrow \{x_i\}, i = 1, \dots, n \\
3. & \text{do } \hat{c} \leftarrow \hat{c} - 1 \\
4. & \text{Find clusters whose merger changes the criterion the least, say, } D_i \text{ and } D_j \\
5. & \text{Merge } D_i \text{ and } D_j \\
6. & \text{until } c = \hat{c} \\
7. & \text{return } c \text{ clusters} \\
8. & \text{end}
\end{array}
$$

We saw earlier that the use of $d_{\text{max}}$ causes the smallest possible stepwise increase in the diameter of the partition. Another simple example is provided by the **sum-of-squared-error criterion function** $J_e$. By an analysis very similar to that used in Section ??, we find that the pair of clusters whose merger increases $J_e$ as little as possible is the pair for which the **distance** 
$$
d_e(D_i, D_j) = \frac{n_i n_j}{n_i + n_j} \left\| m_i - m_j \right\|
$$
is minimum (Equation 78). 

Thus, in selecting clusters to be merged, this criterion takes into account the number of samples in each cluster as well as the distance between clusters. In general, the use of $d_e$ tends to favor growth by merging singletons or small clusters with large clusters over merging medium-sized clusters. While the final partition may not minimize $J_e$, it usually provides a very good starting point for further iterative optimization.


```python
import math

# Function to calculate the Euclidean distance between two points
def euclidean_distance(p1, p2):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(p1, p2)))

# Function to calculate the mean of a list of points
def calculate_mean(cluster):
    n = len(cluster)
    mean_point = [sum(dim) / n for dim in zip(*cluster)]
    return mean_point

# Function to calculate the sum of squared error (Je) between clusters
def sum_of_squared_error(Di, Dj):
    mean_i = calculate_mean(Di)
    mean_j = calculate_mean(Dj)
    
    # Calculate the weighted distance between the means of the two clusters
    ni = len(Di)
    nj = len(Dj)
    
    return (ni * nj) / (ni + nj) * euclidean_distance(mean_i, mean_j)

# Stepwise-optimal hierarchical clustering algorithm
def stepwise_optimal_hierarchical_clustering(data, desired_clusters):
    # Initialize each point as a singleton cluster
    clusters = [[point] for point in data]
    
    # Repeat until the desired number of clusters is reached
    while len(clusters) > desired_clusters:
        min_increase = float('inf')
        merge_i, merge_j = -1, -1
        
        # Find the pair of clusters whose merger minimizes the increase in the criterion function
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                increase = sum_of_squared_error(clusters[i], clusters[j])
                if increase < min_increase:
                    min_increase = increase
                    merge_i, merge_j = i, j
        
        # Merge the two clusters
        new_cluster = clusters[merge_i] + clusters[merge_j]
        clusters = [clusters[k] for k in range(len(clusters)) if k != merge_i and k != merge_j]
        clusters.append(new_cluster)
        
    return clusters

# Example usage:
data = [
    [1.0, 2.0],
    [1.5, 2.5],
    [3.0, 3.0],
    [5.0, 8.0],
    [6.0, 9.0],
    [8.0, 8.0],
    [9.0, 5.0],
    [10.0, 1.0]
]

desired_clusters = 3
final_clusters = stepwise_optimal_hierarchical_clustering(data, desired_clusters)

print("Final clusters:")
for cluster in final_clusters:
    print(cluster)
```

```python
import math
import matplotlib.pyplot as plt

# Function to calculate the Euclidean distance between two points
def euclidean_distance(p1, p2):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(p1, p2)))

# Function to calculate the mean of a list of points
def calculate_mean(cluster):
    n = len(cluster)
    mean_point = [sum(dim) / n for dim in zip(*cluster)]
    return mean_point

# Function to calculate the sum of squared error (Je) between clusters
def sum_of_squared_error(Di, Dj):
    mean_i = calculate_mean(Di)
    mean_j = calculate_mean(Dj)
    
    # Calculate the weighted distance between the means of the two clusters
    ni = len(Di)
    nj = len(Dj)
    
    return (ni * nj) / (ni + nj) * euclidean_distance(mean_i, mean_j)

# Stepwise-optimal hierarchical clustering algorithm
def stepwise_optimal_hierarchical_clustering(data, desired_clusters):
    # Initialize each point as a singleton cluster
    clusters = [[point] for point in data]
    
    # Store the cluster history for plotting
    cluster_history = [clusters.copy()]
    
    # Repeat until the desired number of clusters is reached
    while len(clusters) > desired_clusters:
        min_increase = float('inf')
        merge_i, merge_j = -1, -1
        
        # Find the pair of clusters whose merger minimizes the increase in the criterion function
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                increase = sum_of_squared_error(clusters[i], clusters[j])
                if increase < min_increase:
                    min_increase = increase
                    merge_i, merge_j = i, j
        
        # Merge the two clusters
        new_cluster = clusters[merge_i] + clusters[merge_j]
        clusters = [clusters[k] for k in range(len(clusters)) if k != merge_i and k != merge_j]
        clusters.append(new_cluster)
        
        # Record the cluster arrangement at this step
        cluster_history.append(clusters.copy())
        
    return clusters, cluster_history

# Function to plot clusters
def plot_clusters(data, cluster_history, step=0):
    plt.figure(figsize=(8, 6))
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
    
    # Assign each point to a cluster in the current step
    clusters = cluster_history[step]
    
    # Plot each cluster with a different color
    for i, cluster in enumerate(clusters):
        cluster_points = list(zip(*cluster))
        plt.scatter(cluster_points[0], cluster_points[1], label=f"Cluster {i+1}", color=colors[i % len(colors)], s=50)
    
    plt.title(f"Step {step + 1} Clusters")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend(loc='upper left')
    plt.show()

# Example usage:
data = [
    [1.0, 2.0],
    [1.5, 2.5],
    [3.0, 3.0],
    [5.0, 8.0],
    [6.0, 9.0],
    [8.0, 8.0],
    [9.0, 5.0],
    [10.0, 1.0]
]

desired_clusters = 3
final_clusters, cluster_history = stepwise_optimal_hierarchical_clustering(data, desired_clusters)

# Plot clusters at each step
for step in range(len(cluster_history)):
    plot_clusters(data, cluster_history, step)
```
