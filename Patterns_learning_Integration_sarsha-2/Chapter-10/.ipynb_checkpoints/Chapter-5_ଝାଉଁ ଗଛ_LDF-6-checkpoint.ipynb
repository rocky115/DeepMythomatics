# Linear Programming Algorithms

## 5.10.1 Linear Programming

The Perceptron, relaxation, and Ho-Kashyap procedures are essentially gradient descent methods for solving simultaneous linear inequalities. Linear programming (LP) techniques are procedures for maximizing or minimizing linear functions subject to linear equality or inequality constraints. This suggests that linear inequalities could potentially be solved by using them as constraints in a linear programming problem.

A classical linear programming problem can be stated as follows:

Minimize the linear (scalar) objective function:

$$
z = \alpha^T u \tag{97}
$$

subject to the constraint:

$$
A u \geq \beta \tag{98}
$$

Where:
- $\alpha$ is an $m \times 1$ cost vector,
- $\beta$ is an $l \times 1$ vector,
- $A$ is an $l \times m$ matrix.

The **Simplex algorithm** is the classical iterative procedure for solving this problem. However, for technical reasons, it requires the imposition of one additional constraint: $u \geq 0$. 

This constraint may not be acceptable in some cases (e.g., when solutions require both positive and negative components), but it can be resolved by re-defining the vector $a$ as follows:

$$
a = a^+ - a^- \tag{99}
$$

Where:

$$
a^+ = \frac{1}{2} (|a| + a) \tag{100}
$$

and

$$
a^- = \frac{1}{2} (|a| - a) \tag{101}
$$

Thus, both $a^+$ and $a^-$ are nonnegative. This reformulation allows us to accept the constraint $u \geq 0$, by identifying the components of $u$ with the components of $a^+$ and $a^-$.

## 5.10.2 The Linearly Separable Case

Suppose we have a set of $n$ samples $y_1, y_2, \dots, y_n$ and we want to find a weight vector $a$ that satisfies:

$$
a^T y_i \geq b_i > 0 \quad \text{for all } i \tag{102}
$$

How can we formulate this as a linear programming problem?

One approach is to introduce an artificial variable $\tau \geq 0$ by writing:

$$
a^T y_i + \tau \geq b_i \tag{103}
$$

If $\tau$ is sufficiently large, such as $\tau = \max_i b_i$, the constraints are trivially satisfied (for example, when $a = 0$). However, our goal is to find a solution with $\tau = 0$, which would satisfy the constraints as tightly as possible.

Thus, we are led to consider the following linear programming problem:

Minimize $\tau$ over all values of $\tau$ and $a$ that satisfy the conditions:

$$
a^T y_i \geq b_i \quad \text{and} \quad \tau \geq 0 \tag{104}
$$

### Formal Problem Formulation

The linear programming problem is to minimize the objective function:

$$
z = \alpha^T u \tag{97}
$$

subject to the constraints:

$$
A u \geq \beta \quad \text{and} \quad u \geq 0
$$

Where the matrix $A$ and vectors $\alpha$ and $\beta$ are given by:

$$
A = \begin{bmatrix}
y_1^T & -y_1^T & 1 \\
y_2^T & -y_2^T & 1 \\
\vdots & \vdots & \vdots \\
y_n^T & -y_n^T & 1 \\
\end{bmatrix}, \quad
u = \begin{bmatrix}
a^+ \\
a^- \\
\tau
\end{bmatrix}, \quad
\alpha = \begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix}, \quad
\beta = \begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
$$

Thus, the linear programming problem involves $m = 2d + 1$ variables and $l = n$ constraints, plus the simplex algorithm constraint $u \geq 0$.

### Solution Interpretation

The simplex algorithm will find the minimum value of the objective function:

$$
z = \alpha^T u = \tau
$$

in a finite number of steps, and will exhibit a vector $u^\hat{}$ yielding that value.

- If the minimum value of $\tau$ is zero, the samples are linearly separable, and a solution vector $a^\hat{}$ can be obtained from $u^\hat{}$.
- If the minimum value of $\tau$ is positive, the samples are non-separable, and the resulting $u^\hat{}$ usually doesn't provide a useful approximate solution but does provide proof of non-separability.

This formulation transforms the problem of linear separability into a standard linear programming problem that can be solved using methods like the simplex algorithm.


````python
import numpy as np
from scipy.optimize import linprog

# Define the problem parameters
def linear_programming_separable(y, b):
    """
    Solves the linear programming problem for linearly separable samples.
    
    The problem is:
    Minimize τ
    Subject to: a^T y_i ≥ b_i for all i and τ ≥ 0
    
    :param y: n x d matrix of sample vectors (n samples, d features)
    :param b: n x 1 vector of target values
    :return: The optimal solution vector a (weights), and the minimum value τ
    """
    n, d = y.shape

    # Construct the matrix A (2d x n+1) and the vector b (n)
    # Augment y to include negative components and artificial variable τ
    A = np.zeros((n, 2 * d + 1))
    for i in range(n):
        A[i, :d] = y[i]  # y_i component
        A[i, d:2*d] = -y[i]  # -y_i component
        A[i, 2*d] = 1  # Artificial variable τ
    
    # The objective vector α to minimize is τ (last component)
    c = np.zeros(2 * d + 1)
    c[2 * d] = 1  # Minimizing τ
    
    # The vector b (n) has the target values for each sample
    bounds = [(None, None)] * (2 * d) + [(0, None)]  # 2d components have no bounds, τ ≥ 0
    
    # Solve the linear programming problem
    res = linprog(c, A_ub=-A, b_ub=-b, bounds=bounds, method='simplex')
    
    if res.success:
        # Extract optimal solution vector u = [a+, a-, τ]
        u_opt = res.x
        a_plus = u_opt[:d]
        a_minus = u_opt[d:2*d]
        tau = u_opt[2*d]
        
        # Reconstruct the weight vector a as a+ - a-
        a = a_plus - a_minus
        return a, tau
    else:
        print("Linear programming did not converge.")
        return None, None

# Example Usage
# Define sample points y (n samples, d features) and target values b
y = np.array([[2, 3], [1, 1], [-1, -1], [-2, -3]])  # 4 samples, 2 features
b = np.array([5, 3, 1, -1])  # Target values

# Solve using linear programming
a, tau = linear_programming_separable(y, b)

# Output the results
print("Optimal weight vector a:", a)
print("Optimal value of τ:", tau)
````

````python
import scipy.optimize

# Linear Programming function to find the optimal weights for linearly separable data
def linear_programming_separable(y, b):
    """
    Solves the linear programming problem for linearly separable samples.

    The problem is:
    Minimize τ
    Subject to: a^T y_i ≥ b_i for all i and τ ≥ 0
    
    :param y: List of n x d sample vectors (n samples, d features)
    :param b: List of n target values
    :return: The optimal solution vector a (weights), and the minimum value τ
    """
    n = len(y)  # Number of samples
    d = len(y[0])  # Number of features

    # Construct the matrix A (n x (2d + 1)) and the vector b (n)
    A = []
    for i in range(n):
        row = [0] * (2 * d + 1)
        for j in range(d):
            row[j] = y[i][j]  # y_i component
            row[d + j] = -y[i][j]  # -y_i component
        row[2 * d] = 1  # Artificial variable τ
        A.append(row)

    # The objective vector α to minimize is τ (last component)
    c = [0] * (2 * d + 1)
    c[2 * d] = 1  # Minimizing τ
    
    # Solve the linear programming problem using scipy.optimize.linprog
    result = scipy.optimize.linprog(c, A_ub=-1 * np.array(A), b_ub=-1 * np.array(b), bounds=[(None, None)] * (2 * d) + [(0, None)], method='simplex')
    
    if result.success:
        # Extract optimal solution vector u = [a+, a-, τ]
        u_opt = result.x
        a_plus = u_opt[:d]
        a_minus = u_opt[d:2*d]
        tau = u_opt[2*d]
        
        # Reconstruct the weight vector a as a+ - a-
        a = [a_plus[i] - a_minus[i] for i in range(d)]
        return a, tau
    else:
        print("Linear programming did not converge.")
        return None, None

# Example Usage
# Define sample points y (n samples, d features) and target values b
y = [[2, 3], [1, 1], [-1, -1], [-2, -3]]  # 4 samples, 2 features
b = [5, 3, 1, -1]  # Target values

# Solve using linear programming
a, tau = linear_programming_separable(y, b)

# Output the results
print("Optimal weight vector a:", a)
print("Optimal value of τ:", tau)
````

````python
import matplotlib.pyplot as plt
import random

# Function to generate random data points
def generate_data(n=10, d=2):
    X = [[random.uniform(-10, 10) for _ in range(d)] for _ in range(n)]
    y = [1 if random.random() > 0.5 else -1 for _ in range(n)]
    return X, y

# Plotting function to visualize the data and the separator
def plot_graph(X, y, a=None):
    plt.figure(figsize=(8, 6))
    
    # Plot the points
    for i in range(len(X)):
        if y[i] == 1:
            plt.scatter(X[i][0], X[i][1], color='blue', label='Class 1' if i == 0 else "")
        else:
            plt.scatter(X[i][0], X[i][1], color='red', label='Class -1' if i == 0 else "")

    # Plot the decision boundary if a exists (i.e., the separating hyperplane)
    if a:
        x_vals = [min([point[0] for point in X]), max([point[0] for point in X])]
        y_vals = [-(a[0] * x + a[2]) / a[1] for x in x_vals]  # y = -(a0*x + a2) / a1
        plt.plot(x_vals, y_vals, color='green', label="Separator", linewidth=2)
    
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.legend()
    plt.title('Data Points and Decision Boundary')
    plt.show()

# Linear programming algorithm implementation (simplified)
def linear_programming(X, y):
    # Construct matrix A, vector b and cost vector alpha based on input data
    n = len(X)
    d = len(X[0])
    
    A = []
    b = []
    
    for i in range(n):
        A.append([*X[i], -X[i][0], -X[i][1]])
        b.append(y[i])
    
    # The optimization is to minimize the objective function (we'll focus on τ = 0 for separability)
    # Note: This is a simplified implementation just to generate the boundary and graph
    a_plus = [0] * d
    a_minus = [0] * d
    tau = 0
    for i in range(n):
        if y[i] == 1:
            a_plus = [a_plus[j] + X[i][j] for j in range(d)]
        else:
            a_minus = [a_minus[j] + X[i][j] for j in range(d)]
    
    # A simple separator (this is just an approximation for visualization purposes)
    # The actual linear programming solver would be much more complex (e.g., using Simplex)
    a = [sum(a_plus[j] - a_minus[j] for j in range(d))] + [tau]  # fake simple separator

    return a, 0  # Return the fake separator and 0 iterations

# Main function to run the algorithm and plot results
def main():
    # Generate random data points
    X, y = generate_data(n=10, d=2)
    
    # Plot initial points
    plot_graph(X, y)
    
    # Run the Linear Programming solver (simplified)
    a, iterations = linear_programming(X, y)
    print(f"Separator vector: {a}")
    
    # Plot with the separator (if found)
    plot_graph(X, y, a)

# Run the program
if __name__ == "__main__":
    main()
````

## Minimizing the Perceptron Criterion Function Using Linear Programming

### Introduction

In pattern classification problems, especially when the data is **non-linearly separable**, the goal is often to find a weight vector that classifies as many samples as possible. The **Perceptron criterion function** is used to evaluate the quality of the classifier, and minimizing this criterion can be framed as a **linear programming** (LP) problem.

In this section, we show how the Perceptron criterion function can be transformed into a linear programming formulation, which can then be solved using the **simplex algorithm** or other LP solvers.

## Perceptron Criterion Function

The basic Perceptron criterion function is given by:

$$
J_p(a) = \sum_{y \in Y(a)} (-a^t y)
$$

Where $ Y(a) $ is the set of misclassified samples and $ y $ represents the labels of the training samples.

In practice, this criterion involves minimizing the number of misclassifications. To introduce a margin, we define a modified criterion:

$$
J_p^\prime(a) = \sum_{y \in Y} (b_i - a^t y)
$$

Here, $ b_i $ is the margin for each sample, and the goal is to minimize the errors while maintaining the margin.

## Transforming the Problem into Linear Programming

To convert the non-linear problem into a linear programming problem, we introduce **artificial variables** $ \tau_i $ for each sample. These variables represent the slack in the classification margin. We then aim to minimize the sum of these slack variables.

The objective becomes:

$$
\min \sum_{i=1}^n \tau_i
$$

Subject to the constraints:

$$
\tau_i \geq b_i - a^t y_i \quad \forall i
$$

and

$$
\tau_i \geq 0 \quad \forall i
$$

Thus, the linear programming formulation is:

$$
\min \alpha^t u
$$

Where $ u $ is a vector containing the artificial variables and the weight vector. The constraints are:

$$
A u \geq \beta, \quad u \geq 0
$$

Where $ A $ is a matrix formed by the training data, and $ \beta $ is the corresponding margin vector.

## Linear Programming Formulation

For a dataset with $ n $ samples and $ d $ features, the matrix $ A $ is constructed as:

$$
A = \begin{bmatrix}
y_1^t & -y_1^t & 1 & 0 & \dots & 0 \\
y_2^t & -y_2^t & 0 & 1 & \dots & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots \\
y_n^t & -y_n^t & 0 & 0 & \dots & 1
\end{bmatrix}
$$

The goal is to minimize the objective function:

$$
\min \alpha^t u
$$

Where:

- $ u = [a^+, a^-, \tau_1, \tau_2, \dots, \tau_n] $
- $ \alpha $ is a vector of weights
- $ \beta $ is the margin vector.

### Solution Approach

The **simplex algorithm** is a classical method used to solve such linear programming problems. It iteratively improves the solution by moving along the edges of the feasible region until an optimal solution is found.

If the dataset is **linearly separable**, the optimal solution will have $ \tau_i = 0 $ for all samples, indicating that no slack is required to achieve separation. If the dataset is **not separable**, the optimal solution will have positive values for $ \tau_i $, indicating the necessary slack for the classification.

## Example Code Implementation

Below is a Python implementation of the approach. The code uses a simulated linear programming solver to find the separating vector.


````python

#```python
import random
import matplotlib.pyplot as plt

# Generate random data
def generate_data(n=10, d=2):
    X = [[random.uniform(-10, 10) for _ in range(d)] for _ in range(n)]
    y = [1 if random.random() > 0.5 else -1 for _ in range(n)]
    return X, y

# Plot function for visualizing data
def plot_graph(X, y, a=None):
    plt.figure(figsize=(8, 6))
    for i in range(len(X)):
        plt.scatter(X[i][0], X[i][1], color='blue' if y[i] == 1 else 'red', label='Class 1' if y[i] == 1 else 'Class -1')
    
    if a:
        x_vals = [min([point[0] for point in X]), max([point[0] for point in X])]
        y_vals = [-(a[0] * x + a[2]) / a[1] for x in x_vals]
        plt.plot(x_vals, y_vals, color='green', label="Decision Boundary", linewidth=2)
    
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.show()

# Linear Programming formulation for the Perceptron problem
def linear_programming(X, y):
    n = len(X)
    d = len(X[0])
    
    A = []
    b = []
    
    for i in range(n):
        A.append([*X[i], -X[i][0], -X[i][1]])  # Construct A matrix (X + slack variables)
        b.append(y[i])
    
    # Simulated LP solver for illustration (simplified)
    a_plus = [0] * d
    a_minus = [0] * d
    tau = 0
    
    for i in range(n):
        if y[i] == 1:
            a_plus = [a_plus[j] + X[i][j] for j in range(d)]
        else:
            a_minus = [a_minus[j] + X[i][j] for j in range(d)]
    
    a = [sum(a_plus[j] - a_minus[j] for j in range(d))] + [tau]
    
    return a, 0  # Simulated solution for visualization

# Main execution
def main():
    X, y = generate_data(n=10, d=2)
    plot_graph(X, y)  # Plot initial data
    
    # Run the LP-based Perceptron minimization
    a, iterations = linear_programming(X, y)
    print(f"Found separating vector: {a}")
    
    plot_graph(X, y, a)  # Plot with the separator

# Run
if __name__ == "__main__":
    main()
````

````python
import random
import matplotlib.pyplot as plt

# Generate random data (2D)
def generate_data(n=10, d=2):
    X = [[random.uniform(-10, 10) for _ in range(d)] for _ in range(n)]
    y = [1 if random.random() > 0.5 else -1 for _ in range(n)]
    return X, y

# Plot function for visualizing data and the decision boundary
def plot_graph(X, y, a=None):
    plt.figure(figsize=(8, 6))
    for i in range(len(X)):
        plt.scatter(X[i][0], X[i][1], color='blue' if y[i] == 1 else 'red', label='Class 1' if y[i] == 1 else 'Class -1')
    
    if a:
        x_vals = [min([point[0] for point in X]), max([point[0] for point in X])]
        y_vals = [-(a[0] * x + a[2]) / a[1] for x in x_vals]
        plt.plot(x_vals, y_vals, color='green', label="Decision Boundary", linewidth=2)
    
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.show()

# Linear Programming formulation for the Perceptron problem
def linear_programming(X, y):
    n = len(X)
    d = len(X[0])
    
    # Initialize the variables a+ and a- for positive and negative parts of the weight vector
    a_plus = [0] * d
    a_minus = [0] * d
    tau = 0  # Slack variable
    
    # Construct A matrix and b vector for the constraints
    A = []
    b = []
    
    # Initialize the constraints matrix A and vector b
    for i in range(n):
        A.append([*X[i], -X[i][0], -X[i][1]])  # Construct A matrix (X + slack variables)
        b.append(y[i])
    
    # Simulated LP solver for illustration (simplified)
    for i in range(n):
        if y[i] == 1:
            # For class 1, increase the positive weight vector components
            a_plus = [a_plus[j] + X[i][j] for j in range(d)]
        else:
            # For class -1, increase the negative weight vector components
            a_minus = [a_minus[j] + X[i][j] for j in range(d)]
    
    # Calculate the final weight vector (a = a+ - a-)
    a = [sum(a_plus[j] - a_minus[j] for j in range(d))]
    
    return a, 0  # Return the weight vector as the result

# Main execution for Linear Programming Perceptron Minimization
def main():
    # Generate random 2D data
    X, y = generate_data(n=10, d=2)
    
    # Plot the generated data
    plot_graph(X, y)  # Plot initial data points
    
    # Run the linear programming-based Perceptron minimization
    a, iterations = linear_programming(X, y)
    
    print(f"Found separating vector: {a}")
    
    # Plot the decision boundary (separator)
    plot_graph(X, y, a)  # Plot with the separator

# Run the main function
if __name__ == "__main__":
    main()
````

\hline
\text{Widrow-Hoff (LMS)} & J_s = (a y_i - b_i)^2 & a^{(k+1)} = a^{(k)} + \eta_k (b_k - a^{(k)} y_k) y_k & \eta_k \to 0 \\
\hline
\text{Stochastic} & J_m = E (a^T y - z)^2 & a^{(k+1)} = a^{(k)} + \eta_k y_k & \text{Approximation with } R^{-1}(k+1) = R^{-1}(k) + y_k y_k^T \\

\text{Pseudo-inverse} & J_s = \|Y a - b\|^2 & a = Y^\dagger b & \text{Inverse using pseudo-inverse} \\
\hline
\text{Bayes} & J_s = \|Y a - b\|^2 & b^{(k+1)} = b^{(k)} + \eta (e^{(k)} + |e^{(k)}|) & e^{(k)} = Y a^{(k)} - b^{(k)}, b^{(1)} > 0 \\
\hline
\text{Ho-Kashyap} & J_s = \|Y a - b\|^2 & a^{(k+1)} = Y^\dagger b^{(k)} & \eta_k = |e^{(k)}|^T Y R Y^T |e^{(k)}|, R \text{ positive-definite} \\
\hline
\text{Simplex} & J_p' = \tau_i & \text{Simplex Algorithm} & a y_i + \tau \geq b_i, \tau \geq 0 \\
\hline
\end{array}
$$
