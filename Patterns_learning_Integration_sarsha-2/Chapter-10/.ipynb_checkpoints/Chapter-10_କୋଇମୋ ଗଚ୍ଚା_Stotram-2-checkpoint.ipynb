##  Unsupervised Bayesian Learning

###  The Bayes Classifier

As we saw in Chapter ??, maximum-likelihood methods do not assume the parameter vector $ \theta $ to be random — it is just unknown. In such methods, prior knowledge about the likely values for $ \theta $ is not directly relevant, although in practice such knowledge may be used in choosing good starting points for hill-climbing procedures. 

In this section, however, we shall take a Bayesian approach to unsupervised learning. That is, we shall assume that $ \theta $ is a random variable with a known prior distribution $ p(\theta) $, and we shall use the samples to compute the posterior density $ p(\theta|D) $. Interestingly enough, the analysis will closely parallel the analysis of supervised Bayesian learning (Sect. ??), showing that the two problems are formally very similar.

We begin with an explicit statement of our basic assumptions. We assume that:

1. The number of classes $ c $ is known.
2. The prior probabilities $ P(\omega_j) $ for each class are known, $ j = 1, \dots, c $.
3. The forms for the class-conditional probability densities $ p(x|\omega_j, \theta_j) $ are known, $ j = 1, \dots, c $, but the full parameter vector $ \theta = (\theta_1, \dots, \theta_c) $ is not known.
4. Part of our knowledge about $ \theta $ is contained in a known prior density $ p(\theta) $.
5. The rest of our knowledge about $ \theta $ is contained in a set $ D $ of $ n $ samples $ x_1, \dots, x_n $ drawn independently from the familiar mixture density

$$
p(x|\theta) = \sum_{j=1}^{c} p(x|\omega_j, \theta_j)P(\omega_j).
$$

At this point, we could go directly to the calculation of $ p(\theta|D) $. However, let us first see how this density is used to determine the Bayes classifier. Suppose that a state of nature is selected with probability $ P(\omega_i) $ and a feature vector $ x $ is selected according to the probability law $ p(x|\omega_i, \theta_i) $. To derive the Bayes classifier, we must use all of the information at our disposal to compute the posterior probability $ P(\omega_i|x) $. We exhibit the role of the samples explicitly by writing this as $ P(\omega_i|x, D) $.

By Bayes' formula, we have:

$$
P(\omega_i|x, D) = \frac{p(x|\omega_i, D)P(\omega_i)}{\sum_{j=1}^{c} p(x|\omega_j, D)P(\omega_j)}.
$$

Since the selection of the state of nature $ \omega_i $ was done independently of the previously drawn samples, $ P(\omega_i|D) = P(\omega_i) $, and we obtain:

$$
P(\omega_i|x, D) = \frac{p(x|\omega_i, D)P(\omega_i)}{\sum_{j=1}^{c} p(x|\omega_j, D)P(\omega_j)}.
$$

Central to the Bayesian approach is the introduction of the unknown parameter vector $ \theta $ via:

$$
p(x|\omega_i, D) = \int p(x|\omega_i, \theta_i, D)p(\theta_i|D) d\theta_i.
$$

Since the selection of $ x $ is independent of the samples, we have:

$$
p(x|\omega_i, D) = \int p(x|\omega_i, \theta_i)p(\theta_i|D) d\theta_i.
$$

That is, our best estimate of $ p(x|\omega_i) $ is obtained by averaging $ p(x|\omega_i, \theta_i) $ over $ \theta_i $. Whether or not this is a good estimate depends on the nature of $ p(\theta|D) $, and thus our attention turns at last to that density.

##  Learning the Parameter Vector

We can use Bayes' formula to write:

$$
p(D|\theta)p(\theta) = \frac{p(\theta|D)}{\int p(D|\theta)p(\theta) d\theta},
$$

where the independence of the samples yields the likelihood:

$$
p(D|\theta) = \prod_{k=1}^{n} p(x_k|\theta).
$$

Alternatively, letting $ D_n $ denote the set of $ n $ samples, we can write Eq. (35) in the recursive form:

$$
p(\theta|D_n) = \frac{p(x_n|\theta)p(\theta|D_{n-1})}{\int p(x_n|\theta)p(\theta|D_{n-1}) d\theta}.
$$

These are the basic equations for unsupervised Bayesian learning. Equation (35) emphasizes the relation between the Bayesian and the maximum-likelihood solutions. If $ p(\theta) $ is essentially uniform over the region where $ p(D|\theta) $ peaks, then $ p(\theta|D) $ peaks at the same place. If the only significant peak occurs at $ \theta = \hat{\theta} $, and if the peak is very sharp, then Eq. (32) and Eq. (34) yield:

$$
p(x|\omega_i, D) \approx p(x|\omega_i, \hat{\theta}_i),
$$

and

$$
\frac{p(x|\omega_i, \hat{\theta}_i)P(\omega_i)}{\sum_{j=1}^{c} p(x|\omega_j, \hat{\theta}_j)P(\omega_j)}.
$$

That is, these conditions justify the use of the maximum-likelihood estimate as if it were the true value of $ \theta $ in designing the Bayes classifier.
```python
class BayesianClassifier:
    def __init__(self, c, prior_probs, likelihood_functions, prior_density):
        self.c = c  # number of classes
        self.prior_probs = prior_probs  # P(ω_i) for each class
        self.likelihood_functions = likelihood_functions  # p(x | ω_i, θ_i)
        self.prior_density = prior_density  # p(θ) - prior density for θ

    def compute_posterior(self, x, D_prev):
        """
        Computes the posterior probability p(ω_i | x, D_n) for each class using Bayes' formula:
        p(ω_i | x, D_n) ∝ p(x | ω_i, θ_i) * P(ω_i)
        
        Args:
            x: Current sample
            D_prev: Previous posterior probabilities (for each class)

        Returns:
            posterior: List of posterior probabilities for each class
        """
        # Compute the likelihood for the current sample for each class
        likelihood = self.compute_likelihood(x)
        
        # Compute the unnormalized posterior for each class
        posterior = []
        for i in range(self.c):
            posterior.append(likelihood[i] * D_prev[i])
        
        # Normalize the posterior probabilities to sum to 1
        total = sum(posterior)
        normalized_posterior = [p / total for p in posterior]

        return normalized_posterior

    def compute_likelihood(self, x):
        """
        Computes the likelihood p(x | ω_i, θ_i) for the given sample x.
        Sum of weighted likelihoods for each class.
        
        Args:
            x: Current sample
        
        Returns:
            likelihoods: List of likelihoods for each class
        """
        likelihoods = []
        for i in range(self.c):
            likelihoods.append(self.likelihood_functions[i](x) * self.prior_probs[i])
        return likelihoods

    def classify(self, x, D_prev):
        """
        Classifies the sample x based on the computed posterior probability
        and returns the most probable class.
        
        Args:
            x: Current sample
            D_prev: Previous posterior probabilities (for each class)

        Returns:
            predicted_class: The class with the highest posterior probability
        """
        posterior = self.compute_posterior(x, D_prev)
        
        # Return the class with the highest posterior probability
        predicted_class = max(range(self.c), key=lambda i: posterior[i])
        return predicted_class

    def update_posterior(self, x, D_prev):
        """
        Updates the posterior distribution recursively for each new data point.
        """
        return self.compute_posterior(x, D_prev)


# Example usage

# Assume we have 3 classes (c=3), and predefined prior probabilities and likelihood functions for each class.
prior_probs = [0.3, 0.4, 0.3]  # Prior probabilities P(ω_1), P(ω_2), P(ω_3)

# Define simple likelihood functions for each class (for simplicity, we'll use Gaussian-like distributions)
def likelihood_class_1(x):
    return 0.5  # Example constant likelihood (you can use more sophisticated models)

def likelihood_class_2(x):
    return 0.3  # Example constant likelihood

def likelihood_class_3(x):
    return 0.2  # Example constant likelihood

likelihood_functions = [likelihood_class_1, likelihood_class_2, likelihood_class_3]

# Define a prior density function for θ (assuming a simple uniform distribution over a range of θ values)
def prior_density(theta):
    return 1.0  # Example uniform prior

# Instantiate the classifier
bayesian_classifier = BayesianClassifier(c=3, prior_probs=prior_probs, likelihood_functions=likelihood_functions, prior_density=prior_density)

# Initial posterior is just the prior (before seeing any data)
D_prev = [0.3, 0.4, 0.3]  # Uniform prior for simplicity

# Classify a new sample (e.g., x = some observed data point)
x = 1.0  # Example data point (feature vector)
predicted_class = bayesian_classifier.classify(x, D_prev)

print(f"Predicted class for x={x}: Class {predicted_class + 1}")
```

## Unsupervised Bayesian Learning

###  The Bayes Classifier

As discussed in Chapter ??, maximum-likelihood methods do not assume that the parameter vector $\theta$ is random—it is merely unknown. In such methods, prior knowledge about the likely values for $\theta$ is not directly relevant, although in practice such knowledge can be used to choose good starting points for hill-climbing procedures. In this section, however, we take a Bayesian approach to unsupervised learning. That is, we assume that $\theta$ is a random variable with a known prior distribution $p(\theta)$, and we will use the samples to compute the posterior density $p(\theta|D)$. Interestingly, the analysis closely parallels the analysis of supervised Bayesian learning (Sect. ??), showing that the two problems are formally very similar. 

We begin with the explicit statement of our basic assumptions:

1. The number of classes $c$ is known.
2. The prior probabilities $P(\omega_j)$ for each class are known, $j = 1, \dots, c$.
3. The forms of the class-conditional probability densities $p(x|\omega_j, \theta_j)$ are known, $j = 1, \dots, c$, but the full parameter vector $\theta = (\theta_1, \dots, \theta_c)$ is unknown.
4. Part of our knowledge about $\theta$ is contained in a known prior density $p(\theta)$.
5. The rest of our knowledge about $\theta$ is contained in a set $D = \{x_1, \dots, x_n\}$ of $n$ samples drawn independently from the familiar mixture density
   $$
   p(x|\theta) = \sum_{j=1}^{c} p(x|\omega_j, \theta_j) P(\omega_j).
   $$

At this point, we could go directly to the calculation of $p(\theta|D)$. However, let us first see how this density is used to determine the Bayes classifier.

### Bayes Classifier

Suppose that a state of nature is selected with probability $P(\omega_i)$, and a feature vector $x$ is selected according to the probability law $p(x|\omega_i, \theta_i)$. To derive the Bayes classifier, we must use all of the information at our disposal to compute the posterior probability $P(\omega_i|x)$. We exhibit the role of the samples explicitly by writing this as $P(\omega_i|x, D)$. By Bayes' formula, we have
$$
P(\omega_i|x, D) = \frac{p(x|\omega_i, D) P(\omega_i|D)}{\sum_{j=1}^{c} p(x|\omega_j, D) P(\omega_j|D)}.
$$
Since the selection of the state of nature \(\omega_i\) is done independently of the previously drawn samples, we have \(P(\omega_i|D) = P(\omega_i)\), and the formula simplifies to
$$
P(\omega_i|x, D) = \frac{p(x|\omega_i, D) P(\omega_i)}{\sum_{j=1}^{c} p(x|\omega_j, D) P(\omega_j)}.
$$

### Parameter Estimation and Likelihood

Central to the Bayesian approach is the introduction of the unknown parameter vector $\theta$ via
$$
p(x|\omega_i, D) = \int p(x|\theta, \omega_i, D) p(\theta|\omega_i, D) d\theta.
$$
Since the selection of $x$ is independent of the samples, we have $p(x|\theta, \omega_i, D) = p(x|\omega_i, \theta_i)$, and since knowledge of the state of nature when $x$ is selected tells us nothing about the distribution of $\theta\), we have \(p(\theta|\omega_i, D) = p(\theta|D)$. Thus, we obtain
$$
p(x|\omega_i, D) = \int p(x|\omega_i, \theta_i) p(\theta|D) d\theta.
$$
That is, our best estimate of $p(x|\omega_i)$ is obtained by averaging $p(x|\omega_i, \theta_i)$ over $\theta_i$.

### Bayesian Update of the Parameter Vector

We can use Bayes' formula to write the posterior density of $\theta$ given the data $D$ as
$$
p(\theta|D) = \frac{p(D|\theta) p(\theta)}{\int p(D|\theta) p(\theta) d\theta},
$$
where the likelihood \(p(D|\theta)\) is given by
$$
p(D|\theta) = \prod_{k=1}^{n} p(x_k|\theta).
$$

Alternatively, letting $D_n$ denote the set of $n$ samples, we can write the recursive form
$$
p(\theta|D_n) = \frac{p(x_n|\theta) p(\theta|D_{n-1})}{\int p(x_n|\theta) p(\theta|D_{n-1}) d\theta}.
$$

### Identifiability and Learning

One of the main differences between supervised and unsupervised learning is the issue of identifiability. In supervised learning, the lack of identifiability merely means that we obtain an equivalence class of parameter vectors. Because all of these yield the same component density, lack of identifiability presents no theoretical difficulty.

However, in unsupervised learning, lack of identifiability is much more serious. When $\theta$ cannot be determined uniquely, the mixture cannot be decomposed into its true components. While $p(x|D_n)$ may still converge to $p(x)$, the decomposition $p(x|\omega_i, D_n)$ will not, and a theoretical barrier to learning exists. It is here that a few labeled training samples would be valuable for "decoding" the mixture into its components.

### Computational Complexity

Another challenge in unsupervised learning is computational complexity. In supervised learning, the existence of sufficient statistics allows for solutions that are computationally feasible. In unsupervised learning, however, the samples are drawn from a mixture density, which complicates the process. The likelihood function is given by
$$
p(D|\theta) = \prod_{k=1}^{n} \sum_{j=1}^{c} p(x_k|\omega_j, \theta_j) P(\omega_j),
$$
and this mixture complicates any attempt to find simple exact solutions for $p(\theta|D)$.

An exception to this occurs when the component densities do not overlap, so that as $\theta$ varies, only one term in the mixture density is non-zero. In that case, the likelihood is the product of the non-zero terms and may possess a simple sufficient statistic. However, since this scenario essentially reduces the problem to supervised learning, it is not a significant exception.

### Bayesian Update in Supervised Learning

If we consider the special case where $P(\omega_1) = 1$ and all the other prior probabilities are zero, corresponding to the supervised case where all samples come from Class $\omega_1$, the recursive Bayesian update simplifies to
$$
p(\theta|D_n) \propto p(x_n|\omega_1, \theta_1) p(\theta|D_{n-1}).
$$
The unsupervised case involves multiplying by the mixture density
$$
p(\theta|D_n) \propto \sum_{j=1}^{c} p(x_n|\omega_j, \theta_j) P(\omega_j) p(\theta|D_{n-1}),
$$
which reflects the fact that in the unsupervised case, the sample $x_n$ could have come from any of the classes $\omega_1, \omega_2, \dots, \omega_c$, and thus its influence on the estimate of $\theta$ is distributed across the classes according to their respective probabilities.


```python
import numpy as np
from scipy.stats import norm

# Define the Bayesian Classifier class
class BayesianClassifier:
    def __init__(self, num_classes, prior_probs, class_conditional_densities, prior_theta):
        """
        Initialize the Bayesian Classifier.
        
        :param num_classes: Number of classes c
        :param prior_probs: List of prior probabilities P(omega_j) for each class
        :param class_conditional_densities: List of functions for class-conditional densities p(x|omega_j, theta_j)
        :param prior_theta: Prior for the parameter vector p(theta)
        """
        self.num_classes = num_classes
        self.prior_probs = prior_probs
        self.class_conditional_densities = class_conditional_densities
        self.prior_theta = prior_theta

    def compute_likelihood(self, x, theta):
        """
        Compute the likelihood p(x|theta, omega_j) for each class.
        
        :param x: Input sample
        :param theta: List or array of parameter vectors for each class
        :return: Likelihood for each class
        """
        likelihood = []
        for j in range(self.num_classes):
            # Pass the class-specific parameters theta[j] (which should be a vector of [mean, stddev])
            likelihood.append(self.class_conditional_densities[j](x, theta[j]))
        return np.array(likelihood)

    def compute_posterior(self, x, theta):
        """
        Compute the posterior p(omega_i|x, D) based on Bayes' rule.
        
        :param x: Input sample
        :param theta: List of parameters for each class
        :return: Posterior probabilities for each class
        """
        # Compute likelihoods for each class
        likelihood = self.compute_likelihood(x, theta)
        
        # Update posterior using Bayes' rule
        numerator = likelihood * self.prior_probs
        denominator = np.sum(numerator)
        
        return numerator / denominator

    def classify(self, x, D_prev):
        """
        Classify a new sample x using the posterior probabilities.
        
        :param x: New input sample
        :param D_prev: Set of previously observed data points
        :return: Predicted class for the new sample
        """
        posterior = self.compute_posterior(x, D_prev)
        
        # Choose the class with the maximum posterior probability
        predicted_class = np.argmax(posterior)
        
        return predicted_class


# Example usage

# Define prior probabilities for each class (assumed uniform here)
prior_probs = [0.5, 0.5]

# Define class-conditional densities for each class (assuming normal distributions for simplicity)
def class_conditional_1(x, theta):
    # Expect theta to be a vector [mean, stddev]
    return norm.pdf(x, loc=theta[0], scale=theta[1])

def class_conditional_2(x, theta):
    # Expect theta to be a vector [mean, stddev]
    return norm.pdf(x, loc=theta[0], scale=theta[1])

class_conditional_densities = [class_conditional_1, class_conditional_2]

# Define prior for theta (each class has a [mean, stddev] vector)
prior_theta = [
    [0, 1],  # Class 1: mean=0, stddev=1
    [1, 1]   # Class 2: mean=1, stddev=1
]

# Instantiate the Bayesian classifier
classifier = BayesianClassifier(num_classes=2, 
                                prior_probs=prior_probs, 
                                class_conditional_densities=class_conditional_densities, 
                                prior_theta=prior_theta)

# Initial data (for simplicity, using random data)
D_prev = np.array([0.5, 1.0, 1.5, 2.0])  # This is just a placeholder, not used in likelihood calc

# Classify a new sample (e.g., x = some observed data point)
x = 1.2  # Example data point (feature vector)

# Compute the class probabilities and classify
predicted_class = classifier.classify(x, D_prev)

print(f"Predicted class for x={x}: Class {predicted_class + 1}")
```

## Unsupervised Learning of Gaussian Mixture Models (GMMs)

This notebook demonstrates unsupervised Bayesian learning using a one-dimensional, two-component Gaussian mixture model. The components are:

1. $ p(x|\omega_1) \sim \mathcal{N}(\mu, 1) $ — a Gaussian with mean $\mu$ and variance $1$,
2. 4 p(x|\omega_2, \theta) \sim \mathcal{N}(\theta, 1) $ — a Gaussian with mean $\theta$ (the parameter we want to estimate) and variance $1$.

### Prior Distribution

We assume the prior distribution for $\theta$ is uniform between $a$ and $b$:

$$
p(\theta) = \frac{1}{b - a}, \quad a \leq \theta \leq b
$$

### Likelihood Function

The likelihood function for the second component is a Gaussian distribution:

$$
p(x | \theta) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2} (x - \theta)^2\right)
$$

### Posterior Distribution After One Observation

After observing the first data point $x_1$, the posterior distribution for $\theta$ is given by Bayes' theorem:

$$
p(\theta|x_1) \propto p(x_1|\theta) p(\theta)
$$

Expanding this expression:

$$
p(\theta|x_1) \propto P(\omega_1) \exp\left( -\frac{1}{2}(x_1 - \mu)^2 \right) + P(\omega_2) \exp\left( -\frac{1}{2}(x_1 - \theta)^2 \right)
$$

This is a mixture of two Gaussian distributions: one centered at $\mu$ (due to $P(\omega_1)$) and one centered at $x_1$ (due to $P(\omega_2)$).

### Posterior Distribution After Multiple Observations

With additional data points $ x_2, x_3, \dots, x_n $, the posterior distribution becomes more complex. After observing $n$ data points, the posterior is updated recursively as:

$$
p(\theta | x_1, \dots, x_n) \propto p(x_n|\theta) p(\theta | x_1, \dots, x_{n-1})
$$

As more data is observed, the posterior distribution becomes more concentrated around the true value of $\theta$, and the prior's influence diminishes.

### Numerical Approximation

Since the posterior distribution becomes complicated for larger $n$, numerical methods such as Monte Carlo integration or numerical quadrature can be used to approximate the posterior.

### Convergence of the Posterior Distribution

As the number of data points increases, the posterior distribution converges to an impulse centered around the true value of $\theta$. This process is illustrated below by showing how the posterior distribution becomes more concentrated as \(n\) increases.

### Bayesian vs Maximum Likelihood

In Bayesian learning, the prior plays a crucial role, especially when the number of samples is small. However, as the number of samples increases, the posterior is dominated by the data, and the prior's influence diminishes.

In contrast, maximum likelihood estimation (MLE) focuses solely on the likelihood of the data and ignores the prior. This results in estimates of $\theta$ that depend only on the data points.



```python
### Code Example to Illustrate the Posterior Update

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the prior distribution for theta (uniform prior between a and b)
a, b = -4, 4
prior_theta = np.ones(b - a + 1) / (b - a + 1)  # Uniform prior

# Define the class-conditional densities p(x|omega1) and p(x|omega2, theta)
mu = 2  # Given mean for the first component (omega1)
P_omega1 = 1/3  # Probability of class omega1
P_omega2 = 2/3  # Probability of class omega2

def likelihood_x_given_theta(x, theta):
    """Likelihood p(x|theta) for the second component"""
    return norm.pdf(x, loc=theta, scale=1)

def posterior(theta, x_data):
    """Compute posterior p(theta|x1, ..., xn) based on the data"""
    posterior_values = np.zeros_like(theta)
    for i, t in enumerate(theta):
        posterior_values[i] = (P_omega1 * np.exp(-0.5 * (x_data[0] - mu)**2) +
                               P_omega2 * np.exp(-0.5 * (x_data[0] - t)**2))
    return posterior_values / np.sum(posterior_values)

# Generate synthetic data (e.g., x1 = 3 and x2 = 4)
x_data = [3, 4]

# Set theta values to sample over
theta_values = np.linspace(a, b, 100)

# Compute posterior after observing both x1 and x2
posterior_values = posterior(theta_values, x_data)

# Plot the posterior distribution after observing x1 and x2
plt.plot(theta_values, posterior_values)
plt.title("Posterior Distribution p(theta | x1, x2)")
plt.xlabel("theta")
plt.ylabel("Posterior Probability")
plt.show()
```

```python
import math
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the prior distribution for theta (uniform prior between a and b)
a, b = -4, 4
def prior(theta):
    if a <= theta <= b:
        return 1 / (b - a)
    else:
        return 0

# Define the class-conditional densities p(x|omega1) and p(x|omega2, theta)
mu = 2  # Given mean for the first component (omega1)
P_omega1 = 1/3  # Probability of class omega1
P_omega2 = 2/3  # Probability of class omega2

def likelihood_x_given_theta(x, theta):
    """Likelihood p(x|theta) for the second component"""
    return stats.norm.pdf(x, loc=theta, scale=1)

def posterior(theta_values, x_data):
    """Compute posterior p(theta|x1, ..., xn) based on the data"""
    posterior_values = []
    for t in theta_values:
        likelihood_product = (P_omega1 * math.exp(-0.5 * (x_data[0] - mu) ** 2) +
                              P_omega2 * math.exp(-0.5 * (x_data[0] - t) ** 2))
        posterior_values.append(likelihood_product * prior(t))
    # Normalize the posterior values
    total_posterior = sum(posterior_values)
    return [p / total_posterior for p in posterior_values]

# Generate synthetic data (e.g., x1 = 3 and x2 = 4)
x_data = [3, 4]

# Set theta values to sample over
theta_values = [i * 0.1 for i in range(-40, 41)]  # Values from -4 to 4 with a step size of 0.1

# Compute posterior after observing both x1 and x2
posterior_values = posterior(theta_values, x_data)

# Plot the posterior distribution after observing x1 and x2
plt.plot(theta_values, posterior_values)
plt.title("Posterior Distribution p(theta | x1, x2)")
plt.xlabel("theta")
plt.ylabel("Posterior Probability")
plt.show()
```
