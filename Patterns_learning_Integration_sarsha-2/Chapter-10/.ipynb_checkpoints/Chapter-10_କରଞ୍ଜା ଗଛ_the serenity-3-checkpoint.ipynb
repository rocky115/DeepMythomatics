## Decision-Directed Approximation and Clustering

###  Decision-Directed Approximation

The problem of unsupervised learning can be viewed as estimating parameters of a mixture density. However, neither the maximum-likelihood nor the Bayesian approach provides analytically simple results. The computational requirements of exact solutions grow exponentially with the number of samples. Since unsupervised learning is an important problem, several approximate solutions have been proposed.

### Decision-Directed Approach

A natural approach to unsupervised learning is to use prior information to design a classifier and then use the decisions of this classifier to label the samples. This is known as the **decision-directed approach**. It can be applied in two ways:
1. **Sequential (On-line) Mode**: Update the classifier each time an unlabeled sample is classified.
2. **Batch Mode**: Classify all $n$ samples before updating the classifier.

This process can be repeated until the classifier stabilizes, meaning no further changes occur in the labeling of the samples.

### Limitations of Decision-Directed Approach

- **Classifier Quality**: If the initial classifier is poor, or if the sequence of samples is unfavorable, classification errors can mislead the learning process, leading to suboptimal solutions.
- **Bias in Labeling**: Even if the initial classifier is optimal, the labeling might not match the true class memberships. This is especially true if there is significant overlap between the component densities.

Despite these drawbacks, the simplicity of decision-directed procedures makes them computationally feasible. Even though the solution may be imperfect, it is often better than not having a solution at all. When the parametric assumptions are valid, and there is little overlap between component densities, this method can provide nearly optimal results at a lower computational cost.

##  Data Description and Clustering

In unsupervised learning, our task is often to extract useful information from a set of unlabeled samples. These samples may form clouds of points in a $d$-dimensional space. 

### Basic Data Description with Normal Distribution

If we assume the samples come from a single normal distribution, the most informative characteristics of the data are the **sufficient statistics**: the sample mean and the sample covariance matrix.

1. **Sample Mean** $(\mu$): This locates the center of gravity of the cloud, minimizing the sum of squared distances from $\mu$ to all samples.
2. **Sample Covariance Matrix** $(\Sigma$): This describes how much the data scatters along various directions. For normally distributed data, the cloud of points is typically an ellipsoid, with the sample mean at its center.

If the data is indeed normally distributed, then the cloud will have a simple hyperellipsoidal shape, and the sample mean tends to fall in the region with the highest sample density. However, if the data is not normally distributed, the sample mean and covariance matrix can give a misleading description of the data.

### Limitations of Second-Order Statistics

For datasets where the samples are not normally distributed, second-order statistics (mean and covariance) may fail to reveal the underlying structure. 

For example, Fig.5 shows four different datasets with identical means $(\mu$) and covariance matrices $(\Sigma$). These datasets may have very different structures that are not captured by the sample mean and covariance matrix.

### Mixture of Gaussian Distributions

To better describe data with more complex structures, we can assume that the samples come from a mixture of $c$ normal distributions. This assumption allows us to approximate a wider variety of data distributions. The components of the mixture model represent hyperellipsoidal clouds of different sizes and orientations, which can describe more complex data shapes.

---

## Clustering and Nonparametric Methods

If we lack prior knowledge about the data or its distribution, assuming a parametric form (such as a mixture of Gaussians) may lead to poor results. An alternative approach is to use **nonparametric methods** to estimate the mixture density. These methods provide a more complete description of the data without assuming a specific parametric form.

Nonparametric density estimation methods can identify regions of high local density, which may correspond to meaningful subclasses in the population. These regions are often represented by the **peaks** or **modes** of the estimated density.

### Clustering Procedures

When the goal is to find subclasses in the data, clustering methods can be directly applied. Clustering aims to group data points that have strong internal similarities. Formal clustering methods typically use a criterion function, such as the sum of squared distances from the cluster centers, and seek the grouping that minimizes or maximizes the function.

### Challenges in Clustering

Even though clustering methods can yield useful data descriptions, they often suffer from unmanageable computational problems, especially as the size of the dataset grows. For this reason, many alternative clustering methods have been proposed. These methods are often intuitively appealing and easy to apply, but they may lack rigorous theoretical properties. Nonetheless, they often produce interesting results that can guide more formal and computationally expensive techniques.

---

### Conclusion

In unsupervised learning, the decision-directed approach offers a simple and computationally feasible way to classify unlabeled data, but it has limitations, especially when the initial classifier is not well-designed. More advanced methods, like clustering and nonparametric density estimation, can provide more comprehensive data descriptions, but they also come with their own computational challenges.




##  Similarity Measures

Once we describe the clustering problem as one of finding natural groupings in a set of data, we must define what we mean by a "natural grouping". Specifically, we need to address two key questions:
1. **How should one measure the similarity between samples?**
2. **How should one evaluate a partitioning of a set of samples into clusters?**

This section addresses the first question: How to measure similarity.

## Measuring Similarity with Distance

The most obvious measure of the similarity (or dissimilarity) between two samples is the **distance** between them. A common starting point for clustering is to define a suitable distance function and compute the matrix of distances between all pairs of samples. If distance is a good measure of dissimilarity, we would expect the distance between samples in the same cluster to be significantly less than the distance between samples in different clusters.

### Euclidean Distance Threshold

Suppose we define that two samples belong to the same cluster if the **Euclidean distance** between them is less than some threshold distance \(d_0\). We have:

$$
d(x, x') = \sqrt{\sum_{k=1}^d (x_k - x'_k)^2}
$$

Where $ x $ and $ x' $ are two samples, and $ d $ is the dimensionality of the data. The choice of $ d_0 $ is crucial. If $ d_0 $ is too large, all samples will be assigned to one cluster. If $ d_0 $ is too small, each sample will form a singleton cluster. The threshold $ d_0 $ must be chosen such that it is greater than the typical within-cluster distances and less than typical between-cluster distances, as shown in the figure below.

### Challenges with Euclidean Distance

The choice of Euclidean distance is generally appropriate when the feature space is isotropic, and the data is evenly spread across all directions. However, the results of clustering can be sensitive to the choice of distance metric, especially when data transformations such as scaling or rotation are involved.

#### Effect of Scaling

For example, a simple scaling of the coordinate axes can lead to different clustering results. Figure 10.7 illustrates this problem, where different scalings of the axes affect how clusters are formed. To see this:

1. In the original data, the points are grouped into one cluster (red) and another cluster (gray).
2. After scaling the vertical axis by a factor of 2.0 and shrinking the horizontal axis by a factor of 0.5, the clustering changes.
3. Conversely, scaling the vertical axis by 0.5 and the horizontal axis by 2.0 leads to a different clustering arrangement.

This shows that the Euclidean distance is not invariant to transformations such as scaling.

#### Invariance to Translation and Rotation

To achieve invariance to translations and rotations, one option is to **normalize the data** prior to clustering:
1. **Translation Invariance**: Subtract the mean of each feature.
2. **Scale Invariance**: Divide each feature by its standard deviation (standardize the data).
3. **Rotation Invariance**: Rotate the axes so that they coincide with the eigenvectors of the sample covariance matrix, a transformation often referred to as **principal component analysis (PCA)**.

However, it's important to note that such normalization may not always be desirable. For example, scaling and translation may not be appropriate when the spread of values in the data is due to the presence of subclasses rather than random variation. In such cases, normalization could distort the data and lead to poor clustering results.

### Alternative Distance Metrics

Instead of using Euclidean distance, other distance metrics can be used to handle different types of data relationships. One general class of distance metrics is given by the **Minkowski distance**:

$$
d_q(x, x') = \left( \sum_{k=1}^d |x_k - x'_k|^q \right)^{1/q}
$$

Where $ q \geq 1 $ is a selectable parameter. The special cases of the Minkowski distance are:
- For $ q = 2 $, this reduces to the **Euclidean distance**.
- For $ q = 1 $, this becomes the **Manhattan** or **city block distance**, which is the sum of the absolute distances along each coordinate axis.

It is important to note that only when $ q = 2 $ does the distance remain invariant to arbitrary rotation or translation in feature space.

#### Mahalanobis Distance

An alternative to the Euclidean distance is the **Mahalanobis distance**, which accounts for correlations in the data:

$$
d_M(x, x') = \sqrt{(x - x')^T \Sigma^{-1} (x - x')}
$$

Where $ \Sigma $ is the covariance matrix of the data.

### Similarity Function

In some cases, rather than using a distance, a **similarity function** is used to compare two vectors $ x $ and $ x' $. For example, if the angle between two vectors is a meaningful measure of their similarity, the normalized inner product can be used:

$$
s(x, x') = \frac{x^T x'}{\|x\| \|x'\|}
$$

This similarity function measures how similar two vectors are based on their direction.

---

### Conclusion

In clustering, the measure of similarity between samples is a crucial decision. The Euclidean distance is a commonly used metric, but other distance metrics and similarity functions can be considered based on the problem at hand. It is important to choose a similarity measure that is robust to the inherent transformations in the data and relevant to the clustering task.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data for clustering
n_samples = 300
n_features = 2
n_clusters = 3
data, _ = make_blobs(n_samples=n_samples, centers=n_clusters, n_features=n_features, random_state=42)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Plot the original data
plt.scatter(data[:, 0], data[:, 1], c='gray', label='Original Data', alpha=0.5)
plt.title("Original Data")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Implementing Euclidean Distance
def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))

# Implementing Minkowski Distance
def minkowski_distance(x, y, p=3):
    return np.sum(np.abs(x - y) ** p) ** (1 / p)

# Implementing Mahalanobis Distance
def mahalanobis_distance(x, y, covariance_matrix):
    diff = x - y
    return np.sqrt(np.dot(np.dot(diff.T, np.linalg.inv(covariance_matrix)), diff))

# KMeans Clustering using Euclidean Distance (sklearn automatically uses Euclidean by default)
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(data_scaled)
labels = kmeans.labels_

# Plot the clustering results
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("KMeans Clustering (Euclidean Distance)")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Calculate distances between points using different metrics
x1 = data_scaled[0]
x2 = data_scaled[1]

# Euclidean distance between the first two points
euclid_dist = euclidean_distance(x1, x2)

# Minkowski distance between the first two points (p=3)
minkowski_dist = minkowski_distance(x1, x2, p=3)

# Mahalanobis distance between the first two points (use covariance matrix of the data)
cov_matrix = np.cov(data_scaled.T)
mahalanobis_dist = mahalanobis_distance(x1, x2, covariance_matrix=cov_matrix)

print(f"Euclidean Distance: {euclid_dist}")
print(f"Minkowski Distance (p=3): {minkowski_dist}")
print(f"Mahalanobis Distance: {mahalanobis_dist}")

```
