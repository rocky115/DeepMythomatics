##  The Hessian Matrix

We have shown how the technique of backpropagation can be used to obtain the first derivatives of an error function with respect to the weights in the network. Backpropagation can also be used to evaluate the second derivatives of the error, given by:

$$
\frac{\partial^2 E}{\partial w_{ji} \partial w_{lk}} \tag{5.78}
$$

Note that it is sometimes convenient to consider all of the weight and bias parameters as elements $ w_i $ of a single vector, denoted $ w $, in which case the second derivatives form the elements $ H_{ij} $ of the Hessian matrix $ H $, where $ i, j \in \{1, \dots, W\} $ and $ W $ is the total number of weights and biases. The Hessian plays an important role in many aspects of neural computing, including the following:

1. Several nonlinear optimization algorithms used for training neural networks are based on considerations of the second-order properties of the error surface, which are controlled by the Hessian matrix (Bishop and Nabney, 2008).
2. The Hessian forms the basis of a fast procedure for re-training a feed-forward network following a small change in the training data (Bishop, 1991).
3. The inverse of the Hessian has been used to identify the least significant weights in a network as part of network ‘pruning’ algorithms (Le Cun et al., 1990).
4. The Hessian plays a central role in the Laplace approximation for a Bayesian neural network (see Section 5.7). Its inverse is used to determine the predictive distribution for a trained network, its eigenvalues determine the values of hyperparameters, and its determinant is used to evaluate the model evidence.

Various approximation schemes have been used to evaluate the Hessian matrix for a neural network. However, the Hessian can also be calculated exactly using an extension of the backpropagation technique. An important consideration for many applications of the Hessian is the efficiency with which it can be evaluated. If there are $ W $ parameters (weights and biases) in the network, then the Hessian matrix has dimensions $ W \times W $, and so the computational effort needed to evaluate the Hessian will scale like $ O(W^2) $ for each pattern in the data set. As we shall see, there are efficient methods for evaluating the Hessian whose scaling is indeed $ O(W^2) $.

###  Diagonal Approximation

Some of the applications for the Hessian matrix discussed above require the inverse of the Hessian, rather than the Hessian itself. For this reason, there has been some interest in using a diagonal approximation to the Hessian, in other words, one that simply replaces the off-diagonal elements with zeros, because its inverse is trivial to evaluate. Again, we shall consider an error function $ E $ that consists of a sum of terms, one for each pattern in the data set, so that:

$$
E = \sum_{n} E_n
$$

The Hessian can then be obtained by considering one pattern at a time, and then summing the results over all patterns. From (5.48), the diagonal elements of the Hessian for pattern $ n $ can be written:

$$
\frac{\partial^2 E_n}{\partial w_{ji}^2} = \frac{\partial^2 E_n}{\partial a_j^2} \tag{5.79}
$$

Using (5.48) and (5.49), the second derivatives on the right-hand side of (5.79) can be found recursively using the chain rule of differential calculus to give a backpropagation equation of the form:

$$
\frac{\partial^2 E_n}{\partial a_j^2} = h \left( \frac{\partial^2 E_n}{\partial a_j} \right)^2 w_{kj} w_k + h \left( \frac{\partial E_n}{\partial a_j} \right) w_{kj} \tag{5.80}
$$

If we now neglect off-diagonal elements in the second-derivative terms, we obtain (Becker and Le Cun, 1989; Le Cun et al., 1990):

$$
\frac{\partial^2 E_n}{\partial a_j^2} = h \left( \frac{\partial^2 E_n}{\partial a_j} \right)^2 w_{kj}^2 + h \left( \frac{\partial E_n}{\partial a_j} \right) w_{kj} \tag{5.81}
$$

Note that the number of computational steps required to evaluate this approximation is $ O(W) $, where $ W $ is the total number of weight and bias parameters in the network, compared with $ O(W^2) $ for the full Hessian. 

The major problem with diagonal approximations, however, is that in practice the Hessian is typically found to be strongly nondiagonal, and so these approximations, which are driven mainly by computational convenience, must be treated with care.

###  Outer Product Approximation

When neural networks are applied to regression problems, it is common to use a sum-of-squares error function of the form:

$$
E = \frac{1}{2} \sum_{n=1}^{N} (y_n - t_n)^2 \tag{5.82}
$$

where we have considered the case of a single output in order to keep the notation simple (the extension to several outputs is straightforward). We can then write the Hessian matrix in the form:

$$
H = \nabla \nabla E = \sum_{n=1}^{N} \nabla y_n \nabla y_n + (y_n - t_n) \nabla \nabla y_n \tag{5.83}
$$

If the network has been trained on the data set, and its outputs $ y_n $ happen to be very close to the target values $ t_n $, then the second term in (5.83) will be small and can be neglected. More generally, however, it may be appropriate to neglect this term by the following argument. Recall from Section 1.5.5 that the optimal function that minimizes a sum-of-squares loss is the conditional average of the target data. The quantity $ (y_n - t_n) $ is then a random variable with zero mean. If we assume that its value is uncorrelated with the value of the second derivative term on the right-hand side of (5.83), then the whole term will average to zero in the summation over $ n $. By neglecting the second term in (5.83), we arrive at the Levenberg–Marquardt approximation or outer product approximation (because the Hessian matrix is built up from a sum of outer products of vectors), given by:

$$
H \approx \sum_{n=1}^{N} b_n b_n^T \tag{5.84}
$$

where $ b_n = \nabla y_n = \nabla a_n $, because the activation function for the output units is simply the identity. Evaluation of the outer product approximation for the Hessian is straightforward as it only involves first derivatives of the error function, which can be evaluated efficiently in $ O(W) $ steps using standard backpropagation. The elements of the matrix can then be found in $ O(W^2) $ steps by simple multiplication.

It is important to emphasize that this approximation is only likely to be valid for a network that has been trained appropriately, and that for a general network mapping, the second derivative terms on the right-hand side of (5.83) will typically not be negligible. 

In the case of the cross-entropy error function for a network with logistic sigmoid output-unit activation functions, the corresponding approximation is given by:

$$
H \approx \sum_{n=1}^{N} y_n (1 - y_n) b_n b_n^T \tag{5.85}
$$

An analogous result can be obtained for multiclass networks having softmax output-unit activation functions.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)

def mse_loss_derivative(y_pred, y_true):
    return 2 * (y_pred - y_true) / len(y_true)

# Simple 2-layer Neural Network forward and backward propagation
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases
        self.W1 = np.random.randn(self.input_size, self.hidden_size)
        self.b1 = np.zeros(self.hidden_size)
        self.W2 = np.random.randn(self.hidden_size, self.output_size)
        self.b2 = np.zeros(self.output_size)
    
    def forward(self, X):
        self.hidden_input = np.dot(X, self.W1) + self.b1
        self.hidden_output = sigmoid(self.hidden_input)
        self.output_input = np.dot(self.hidden_output, self.W2) + self.b2
        self.output = sigmoid(self.output_input)
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Compute gradients for output layer
        output_error = mse_loss_derivative(self.output, y)
        output_delta = output_error * sigmoid_derivative(self.output_input)
        
        # Compute gradients for hidden layer
        hidden_error = output_delta.dot(self.W2.T)
        hidden_delta = hidden_error * sigmoid_derivative(self.hidden_input)
        
        # Compute weight and bias gradients
        dW2 = self.hidden_output.T.dot(output_delta)
        db2 = np.sum(output_delta, axis=0)
        dW1 = X.T.dot(hidden_delta)
        db1 = np.sum(hidden_delta, axis=0)
        
        # Update weights and biases
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
        
        # Return the gradients
        return dW1, db1, dW2, db2

# Generate some toy data for training
X_train = np.random.randn(100, 2)  # 100 samples, 2 features
y_train = (np.sum(X_train, axis=1) > 0).astype(np.float32).reshape(-1, 1)  # XOR-like data

# Compute the Hessian matrix for the network
def compute_hessian(nn, X, y):
    # Perform forward pass
    y_pred = nn.forward(X)
    
    # Perform backward pass to compute gradients
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Compute second derivatives (Hessian)
    H = np.zeros((nn.input_size + nn.hidden_size + nn.output_size, nn.input_size + nn.hidden_size + nn.output_size))
    
    # Diagonal Hessian approximation
    H_diag = np.zeros_like(H)
    for i in range(nn.input_size):
        for j in range(nn.hidden_size):
            H_diag[i, j] = np.sum(dW1 ** 2)  # Approximation for diagonal
    
    # Outer product approximation
    H_outer = np.outer(dW1.flatten(), dW1.flatten())  # Using the outer product of gradients
    
    return H, H_diag, H_outer

# Create a simple neural network instance
nn = SimpleNN(input_size=2, hidden_size=4, output_size=1)

# Compute Hessian and its approximations
H, H_diag, H_outer = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Matrix and its Approximations
plt.figure(figsize=(12, 6))

# Full Hessian
plt.subplot(1, 3, 1)
plt.imshow(H, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Full Hessian")

# Diagonal Approximation
plt.subplot(1, 3, 2)
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Diagonal Approximation")

# Outer Product Approximation
plt.subplot(1, 3, 3)
plt.imshow(H_outer, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Outer Product Approximation")

plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return sum((y_pred[i] - y_true[i]) ** 2 for i in range(len(y_pred))) / len(y_pred)

def mse_loss_derivative(y_pred, y_true):
    # Compare individual scalar values in nested lists
    return [2 * (y_pred[i][0] - y_true[i][0]) / len(y_true) for i in range(len(y_pred))]

# Simple 2-layer Neural Network forward and backward propagation
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases with random values
        self.W1 = [[random.gauss(0, 1) for _ in range(self.hidden_size)] for _ in range(self.input_size)]
        self.b1 = [0] * self.hidden_size
        self.W2 = [[random.gauss(0, 1) for _ in range(self.output_size)] for _ in range(self.hidden_size)]
        self.b2 = [0] * self.output_size
    
    def forward(self, X):
        # Forward pass through hidden layer
        self.hidden_input = [
            [sum(X[k][i] * self.W1[i][j] for i in range(self.input_size)) + self.b1[j] for j in range(self.hidden_size)]
            for k in range(len(X))  # Iterate over each sample in the dataset
        ]
        self.hidden_output = [
            [sigmoid(x) for x in sample_input] for sample_input in self.hidden_input
        ]
        
        # Forward pass through output layer
        self.output_input = [
            [sum(self.hidden_output[k][i] * self.W2[i][j] for i in range(self.hidden_size)) + self.b2[j] for j in range(self.output_size)]
            for k in range(len(X))  # Iterate over each sample in the dataset
        ]
        
        self.output = [
            [sigmoid(x) for x in sample_input] for sample_input in self.output_input
        ]
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Compute gradients for output layer
        output_error = mse_loss_derivative(self.output, y)
        output_delta = [
            [output_error[k][i] * sigmoid_derivative(self.output_input[k][i]) for i in range(self.output_size)]
            for k in range(len(X))
        ]
        
        # Compute gradients for hidden layer
        hidden_error = [
            [sum(output_delta[k][j] * self.W2[i][j] for j in range(self.output_size)) for i in range(self.hidden_size)]
            for k in range(len(X))
        ]
        hidden_delta = [
            [hidden_error[k][i] * sigmoid_derivative(self.hidden_input[k][i]) for i in range(self.hidden_size)]
            for k in range(len(X))
        ]
        
        # Compute weight and bias gradients
        dW2 = [
            [sum(self.hidden_output[k][i] * output_delta[k][j] for k in range(len(X))) for i in range(self.hidden_size)]
            for j in range(self.output_size)
        ]
        db2 = [sum(output_delta[k][j] for k in range(len(X))) for j in range(self.output_size)]
        dW1 = [
            [sum(X[k][i] * hidden_delta[k][j] for k in range(len(X))) for i in range(self.input_size)]
            for j in range(self.hidden_size)
        ]
        db1 = [sum(hidden_delta[k][j] for k in range(len(X))) for j in range(self.hidden_size)]
        
        # Update weights and biases
        self.W2 = [
            [self.W2[i][j] - learning_rate * dW2[j][i] for j in range(self.output_size)]
            for i in range(self.hidden_size)
        ]
        self.b2 = [self.b2[i] - learning_rate * db2[i] for i in range(self.output_size)]
        self.W1 = [
            [self.W1[i][j] - learning_rate * dW1[j][i] for j in range(self.hidden_size)]
            for i in range(self.input_size)
        ]
        self.b1 = [self.b1[i] - learning_rate * db1[i] for i in range(self.hidden_size)]
        
        return dW1, db1, dW2, db2

# Generate some toy data for training
def generate_data(n_samples, input_size):
    X = [[random.gauss(0, 1) for _ in range(input_size)] for _ in range(n_samples)]
    y = [[1 if sum(X[i]) > 0 else 0] for i in range(n_samples)]  # XOR-like data
    return X, y

# Compute the Hessian matrix for the network
def compute_hessian(nn, X, y):
    # Perform forward pass
    y_pred = nn.forward(X)
    
    # Perform backward pass to compute gradients
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Flatten the gradients to compute the Hessian
    W1_flat = [item for sublist in nn.W1 for item in sublist]
    W2_flat = [item for sublist in nn.W2 for item in sublist]
    
    # Hessian matrix of second derivatives (approximation)
    H = [[0] * (len(W1_flat) + len(W2_flat)) for _ in range(len(W1_flat) + len(W2_flat))]
    
    # Diagonal Hessian approximation (second derivatives of loss w.r.t parameters)
    H_diag = [[0] * (len(W1_flat) + len(W2_flat)) for _ in range(len(W1_flat) + len(W2_flat))]
    for i in range(len(W1_flat)):
        H_diag[i][i] = sum(dW1[i] ** 2)  # Approximation for diagonal
    
    # Outer product approximation (for each layer)
    H_outer = [[dW1[i] * dW2[j] for j in range(len(W2_flat))] for i in range(len(W1_flat))]
    
    return H, H_diag, H_outer

# Create a simple neural network instance
input_size = 2
hidden_size = 4
output_size = 1
nn = SimpleNN(input_size, hidden_size, output_size)

# Generate training data
X_train, y_train = generate_data(100, input_size)

# Compute Hessian and its approximations
H, H_diag, H_outer = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Matrix and its Approximations
plt.figure(figsize=(12, 6))

# Full Hessian (this is just a placeholder, no computation for full Hessian in this version)
plt.subplot(1, 3, 1)
plt.imshow(H, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Full Hessian")

# Diagonal Approximation
plt.subplot(1, 3, 2)
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Diagonal Approximation")

# Outer Product Approximation
plt.subplot(1, 3, 3)
plt.imshow(H_outer, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Outer Product Approximation")

plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return sum((y_pred[i] - y_true[i]) ** 2 for i in range(len(y_pred))) / len(y_pred)

def mse_loss_derivative(y_pred, y_true):
    return [(y_pred[i] - y_true[i]) * 2 / len(y_true) for i in range(len(y_pred))]

# Simple Neural Network
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases with random values
        self.W1 = [[random.gauss(0, 1) for _ in range(hidden_size)] for _ in range(input_size)]
        self.b1 = [0] * hidden_size
        self.W2 = [[random.gauss(0, 1) for _ in range(output_size)] for _ in range(hidden_size)]
        self.b2 = [0] * output_size
    
    def forward(self, X):
        # Forward pass through hidden layer
        self.hidden_input = [[sum(X[k][i] * self.W1[i][j] for i in range(self.input_size)) + self.b1[j] for j in range(self.hidden_size)] for k in range(len(X))]
        self.hidden_output = [[sigmoid(val) for val in sample] for sample in self.hidden_input]
        
        # Forward pass through output layer
        self.output_input = [[sum(self.hidden_output[k][i] * self.W2[i][j] for i in range(self.hidden_size)) + self.b2[j] for j in range(self.output_size)] for k in range(len(X))]
        self.output = [[sigmoid(val) for val in sample] for sample in self.output_input]
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Backpropagation
        output_error = mse_loss_derivative(self.output, y)
        output_delta = [[output_error[k] * sigmoid_derivative(self.output_input[k][0]) for _ in range(self.output_size)] for k in range(len(X))]
        
        hidden_error = [[sum(output_delta[k][j] * self.W2[i][j] for j in range(self.output_size)) for i in range(self.hidden_size)] for k in range(len(X))]
        hidden_delta = [[hidden_error[k][i] * sigmoid_derivative(self.hidden_input[k][i]) for i in range(self.hidden_size)] for k in range(len(X))]
        
        # Gradient descent update for weights and biases
        dW2 = [[sum(self.hidden_output[k][i] * output_delta[k][j] for k in range(len(X))) for i in range(self.hidden_size)] for j in range(self.output_size)]
        db2 = [sum(output_delta[k][j] for k in range(len(X))) for j in range(self.output_size)]
        dW1 = [[sum(X[k][i] * hidden_delta[k][j] for k in range(len(X))) for i in range(self.input_size)] for j in range(self.hidden_size)]
        db1 = [sum(hidden_delta[k][j] for k in range(len(X))) for j in range(self.hidden_size)]
        
        # Update weights and biases
        self.W2 = [[self.W2[i][j] - learning_rate * dW2[j][i] for j in range(self.output_size)] for i in range(self.hidden_size)]
        self.b2 = [self.b2[i] - learning_rate * db2[i] for i in range(self.output_size)]
        self.W1 = [[self.W1[i][j] - learning_rate * dW1[j][i] for j in range(self.hidden_size)] for i in range(self.input_size)]
        self.b1 = [self.b1[i] - learning_rate * db1[i] for i in range(self.hidden_size)]
        
        return dW1, db1, dW2, db2

# Generate some toy data for training
def generate_data(n_samples, input_size):
    X = [[random.gauss(0, 1) for _ in range(input_size)] for _ in range(n_samples)]
    y = [[1 if sum(X[i]) > 0 else 0] for i in range(n_samples)]  # Simple decision boundary
    return X, y

# Simplified Hessian approximation
def compute_hessian(nn, X, y):
    y_pred = nn.forward(X)
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Diagonal approximation for Hessian
    H_diag = [[dW1[i][j] ** 2 + dW2[i][j] ** 2 for j in range(len(dW1[0]))] for i in range(len(dW1))]
    return H_diag

# Create a simple neural network instance
input_size = 2
hidden_size = 4
output_size = 1
nn = SimpleNN(input_size, hidden_size, output_size)

# Generate training data
X_train, y_train = generate_data(100, input_size)

# Compute Hessian approximation
H_diag = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Diagonal
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Hessian Diagonal Approximation")
plt.show()
```

## Inverse Hessian and Finite Differences

This notebook discusses two techniques for working with the Hessian matrix in neural network training:
1. **Inverse Hessian using the outer-product approximation**.
2. **Finite difference approximations for second derivatives**.

---

## 1. Inverse Hessian

We begin by recalling the outer-product approximation for the Hessian matrix:

$$
H_N = \sum_{n=1}^{N} b_n b_n^T
$$

where $ b_n \equiv \nabla_w a_n $ is the contribution to the gradient of the output unit activation arising from data point $ n $.

### Sequential Procedure for Building the Inverse Hessian

Suppose the inverse Hessian has already been computed using the first $ L $ data points. When adding the contribution from data point $ L+1 $, the updated Hessian is given by:

$$
H_{L+1} = H_L + b_{L+1} b_{L+1}^T
$$

Using the **Woodbury identity**, we can directly compute the inverse of $ H_{L+1} $. The identity is:

$$
(M + vv^T)^{-1} = M^{-1} - \frac{M^{-1} v v^T M^{-1}}{1 + v^T M^{-1} v}
$$

By identifying $ H_L $ with $ M $ and $ b_{L+1} $ with $ v $, the inverse Hessian update becomes:

$$
H_{L+1}^{-1} = H_L^{-1} - \frac{H_L^{-1} b_{L+1} b_{L+1}^T H_L^{-1}}{1 + b_{L+1}^T H_L^{-1} b_{L+1}}
$$

### Initialization

The initial matrix $ H_0 $ is chosen as:

$$
H_0 = \alpha I
$$

where $ \alpha $ is a small constant to ensure numerical stability. The algorithm ultimately computes the inverse of $ H + \alpha I $, but the results are generally not sensitive to the specific value of $\alpha $.

This sequential procedure processes the entire dataset in a single pass, making it computationally efficient.

---

## 2. Finite Difference Approximations for Second Derivatives

For cases where direct computation of the second derivatives is infeasible, **finite difference methods** can be used. The second derivative of the error function is approximated as:

$$
\frac{\partial^2 E}{\partial w_{ji} \partial w_{lk}} = \frac{1}{4\epsilon^2} \big[
    E(w_{ji} + \epsilon, w_{lk} + \epsilon) - E(w_{ji} + \epsilon, w_{lk} - \epsilon)
$$
$$
    - E(w_{ji} - \epsilon, w_{lk} + \epsilon) + E(w_{ji} - \epsilon, w_{lk} - \epsilon)
\big] + \mathcal{O}(\epsilon^2)
$$

where $ \epsilon $ is a small perturbation.

### Practical Notes

- The accuracy of this approximation is limited by numerical precision, particularly for very small values of $ \epsilon $.
- Despite this limitation, finite difference methods are useful for verifying implementations of the Hessian computation.


```python
import numpy as np

# Initialize a small neural network model with weights and biases
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.random.randn(hidden_size)
        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.random.randn(output_size)

    def forward(self, X):
        self.hidden_input = X @ self.W1 + self.b1
        self.hidden_output = 1 / (1 + np.exp(-self.hidden_input))  # Sigmoid activation
        self.output_input = self.hidden_output @ self.W2 + self.b2
        self.output = 1 / (1 + np.exp(-self.output_input))  # Sigmoid activation
        return self.output

    def loss(self, y_pred, y_true):
        return 0.5 * np.sum((y_pred - y_true) ** 2)

# Sequential Inverse Hessian Computation using the Woodbury Identity
def compute_inverse_hessian(X, nn, alpha=1e-4):
    n_samples, input_size = X.shape
    hidden_size = nn.W1.shape[1]
    output_size = nn.W2.shape[1]

    # Total number of parameters (weights and biases)
    n_params = nn.W1.size + nn.b1.size + nn.W2.size + nn.b2.size

    # Initial inverse Hessian (alpha * Identity matrix)
    H_inv = np.eye(n_params) * alpha

    for x in X:
        x = x.reshape(1, -1)
        nn.forward(x)

        # Flatten weights and biases to create a single gradient vector
        b = np.hstack((
            nn.W1.flatten(),
            nn.b1.flatten(),
            nn.W2.flatten(),
            nn.b2.flatten()
        )).reshape(-1, 1)

        H_inv_b = H_inv @ b
        H_inv -= (H_inv_b @ H_inv_b.T) / (1 + b.T @ H_inv_b)

    return H_inv

# Example usage
if __name__ == "__main__":
    # Create a small neural network
    input_size = 2
    hidden_size = 4
    output_size = 1
    nn = SimpleNN(input_size, hidden_size, output_size)

    # Generate dummy data
    X = np.random.randn(10, input_size)
    y = (np.sum(X, axis=1) > 0).astype(float).reshape(-1, 1)

    # Compute inverse Hessian
    H_inv = compute_inverse_hessian(X, nn)
    print("Inverse Hessian:")
    print(H_inv)
```

```python
import random
import math

# Define a small neural network model with weights and biases
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.W1 = [[random.random() for _ in range(hidden_size)] for _ in range(input_size)]
        self.b1 = [random.random() for _ in range(hidden_size)]
        self.W2 = [[random.random() for _ in range(output_size)] for _ in range(hidden_size)]
        self.b2 = [random.random() for _ in range(output_size)]

    def forward(self, X):
        self.hidden_input = [
            [sum(X[i] * self.W1[i][j] for i in range(len(self.W1))) + self.b1[j] for j in range(len(self.b1))]
        ]
        self.hidden_output = [
            [1 / (1 + math.exp(-x)) for x in self.hidden_input[0]]
        ]
        self.output_input = [
            [sum(self.hidden_output[0][i] * self.W2[i][j] for i in range(len(self.W2))) + self.b2[j] for j in range(len(self.b2))]
        ]
        self.output = [
            [1 / (1 + math.exp(-x)) for x in self.output_input[0]]
        ]
        return self.output

# Helper functions
def identity_matrix(size, alpha=0):
    """Create an identity matrix scaled by alpha."""
    return [[alpha if i == j else 0 for j in range(size)] for i in range(size)]

def mat_vec_mul(matrix, vector):
    """Multiply a matrix by a vector."""
    return [sum(matrix[i][j] * vector[j] for j in range(len(vector))) for i in range(len(matrix))]

def outer_product(vec1, vec2):
    """Compute the outer product of two vectors."""
    return [[vec1[i] * vec2[j] for j in range(len(vec2))] for i in range(len(vec1))]

def mat_add(mat1, mat2):
    """Add two matrices element-wise."""
    return [[mat1[i][j] + mat2[i][j] for j in range(len(mat1[0]))] for i in range(len(mat1))]

def mat_sub(mat1, mat2):
    """Subtract two matrices element-wise."""
    return [[mat1[i][j] - mat2[i][j] for j in range(len(mat1[0]))] for i in range(len(mat1))]

def scalar_mat_mul(scalar, matrix):
    """Multiply a matrix by a scalar."""
    return [[scalar * matrix[i][j] for j in range(len(matrix[0]))] for i in range(len(matrix))]

# Compute Inverse Hessian using Woodbury identity
def compute_inverse_hessian(X, nn, alpha=1e-4):
    # Total number of parameters (flatten weights and biases)
    n_params = len(nn.W1) * len(nn.W1[0]) + len(nn.b1) + len(nn.W2) * len(nn.W2[0]) + len(nn.b2)

    # Initialize the inverse Hessian as alpha times the identity matrix
    H_inv = identity_matrix(n_params, alpha)

    for x in X:
        x = x  # Input data sample
        nn.forward(x)

        # Flatten all weights and biases into a single vector
        b = []
        for row in nn.W1:
            b.extend(row)
        b.extend(nn.b1)
        for row in nn.W2:
            b.extend(row)
        b.extend(nn.b2)

        # Compute the update using Woodbury identity
        H_inv_b = mat_vec_mul(H_inv, b)
        numerator = outer_product(H_inv_b, H_inv_b)
        denominator = 1 + sum(H_inv_b[i] * b[i] for i in range(len(b)))

        update = scalar_mat_mul(1 / denominator, numerator)
        H_inv = mat_sub(H_inv, update)

    return H_inv

# Example usage
if __name__ == "__main__":
    # Create a small neural network
    input_size = 2
    hidden_size = 4
    output_size = 1
    nn = SimpleNN(input_size, hidden_size, output_size)

    # Generate dummy data
    X = [[random.random() for _ in range(input_size)] for _ in range(10)]
    y = [[1 if sum(sample) > 0 else 0] for sample in X]

    # Compute inverse Hessian
    H_inv = compute_inverse_hessian(X, nn)
    print("Inverse Hessian:")
    for row in H_inv:
        print(row)
```

## Hessian Matrix Evaluation in Neural Networks

In this notebook, we explore methods to compute the Hessian matrix in neural networks, including:
1. **Central Differences for Numerical Approximation**.
2. **Exact Evaluation of the Hessian Using Backpropagation**.

---

## Central Differences for Numerical Approximation

Using symmetrical central differences, we compute the second derivative of the error function $ E $ with respect to two weights $ w_{ji} $ and $ w_{lk} $:

$$
\frac{\partial^2 E}{\partial w_{ji} \partial w_{lk}} = 
\frac{1}{4\epsilon} 
\Big[
E(w_{ji} + \epsilon, w_{lk} + \epsilon) - 
E(w_{ji} + \epsilon, w_{lk} - \epsilon) - 
E(w_{ji} - \epsilon, w_{lk} + \epsilon) + 
E(w_{ji} - \epsilon, w_{lk} - \epsilon)
\Big] + O(\epsilon^2).
$$

This approach requires $ O(W^3) $ operations, where $ W $ is the total number of weights in the network. Although computationally expensive, it is a useful way to validate the implementation of backpropagation.

---

### Optimized Central Differences Using Backpropagation

We can improve efficiency by applying central differences directly to the first derivatives of the error function, which are computed using backpropagation:

$$
\frac{\partial^2 E}{\partial w_{ji} \partial w_{lk}} = 
\frac{1}{2\epsilon} 
\Big[
\frac{\partial E}{\partial w_{ji}}(w_{lk} + \epsilon) - 
\frac{\partial E}{\partial w_{ji}}(w_{lk} - \epsilon)
\Big] + O(\epsilon^2).
$$

This method scales as $ O(W^2) $, making it significantly more efficient than the previous approach.

---

## Exact Evaluation of the Hessian

The exact Hessian can be evaluated for networks of arbitrary feed-forward topology by extending the backpropagation method. This is efficient and scales as $ O(W^2) $. Below, we focus on a network with two layers of weights.

---

### Definitions

1. The derivative of the error function $ E_n $ for a single data point $ n $ with respect to the activation $ a_k $ of the output unit:
   $$
   \delta_k = \frac{\partial E_n}{\partial a_k}.
   $$

2. The second derivative of the error:
   $$
   M_{kk'} = \frac{\partial^2 E_n}{\partial a_k \partial a_{k'}}.
   $$

---

### Hessian Matrix Structure

The Hessian matrix consists of three blocks:

#### 1. Weights in the Second Layer
The second derivative of $ E_n $ with respect to two weights $ w_{kj} $ and $ w_{k'j'} $ in the second layer:

$$
\frac{\partial^2 E_n}{\partial w_{kj} \partial w_{k'j'}} = z_j z_{j'} M_{kk'}.
$$

#### 2. Weights in the First Layer
The second derivative with respect to two weights $ w_{ji} $ and $ w_{j'i'} $ in the first layer:

$$
\frac{\partial^2 E_n}{\partial w_{ji} \partial w_{j'i'}} = 
x_i x_{i'} h'(a_{j'}) I_{jj'} w_{k'j'} \delta_k
+ 
x_i x_{i'} h'(a_{j'}) h'(a_j) w_{k'j'} w_{kj} M_{kk'}.
$$

Here:
- $ h'(a_j) $ is the derivative of the activation function $ h(a_j) $.
- $ I_{jj'} $ is the $ j, j' $-th element of the identity matrix.

#### 3. Weights in Different Layers
The second derivative with respect to a weight in the first layer $ w_{ji} $ and a weight in the second layer $ w_{kj'} $:

$$
\frac{\partial^2 E_n}{\partial w_{ji} \partial w_{kj'}} = 
x_i h'(a_{j'}) \delta_k I_{jj'} + 
z_j w_{k'j'} M_{kk'}.
$$

---

### Notes:
1. If either weight is a bias term, the corresponding activation is set to 1.
2. Extensions to networks with skip-layer connections are straightforward.

---



```python
import math

# Activation function and its derivative
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

# Simple neural network class
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases randomly
        self.W1 = [[random.uniform(-1, 1) for _ in range(hidden_size)] for _ in range(input_size)]
        self.b1 = [0] * hidden_size
        self.W2 = [[random.uniform(-1, 1) for _ in range(output_size)] for _ in range(hidden_size)]
        self.b2 = [0] * output_size

    def forward(self, X):
        # Hidden layer computations
        self.hidden_input = [
            sum(X[i] * self.W1[i][j] for i in range(len(X))) + self.b1[j]
            for j in range(len(self.b1))
        ]
        self.hidden_output = [sigmoid(x) for x in self.hidden_input]

        # Output layer computations
        self.output_input = [
            sum(self.hidden_output[i] * self.W2[i][j] for i in range(len(self.hidden_output))) + self.b2[j]
            for j in range(len(self.b2))
        ]
        self.output = [sigmoid(x) for x in self.output_input]
        return self.output

    def backward(self, X, y):
        # Gradients with respect to the output layer
        output_error = [self.output[k] - y[k] for k in range(len(y))]
        self.delta_output = [
            output_error[k] * sigmoid_derivative(self.output_input[k]) for k in range(len(self.output_input))
        ]

        # Gradients with respect to the hidden layer
        self.delta_hidden = [
            sigmoid_derivative(self.hidden_input[j]) * sum(
                self.delta_output[k] * self.W2[j][k] for k in range(len(self.delta_output))
            )
            for j in range(len(self.hidden_input))
        ]

        # Compute weight and bias gradients
        self.dW2 = [
            [self.delta_output[k] * self.hidden_output[j] for k in range(len(self.delta_output))]
            for j in range(len(self.hidden_output))
        ]
        self.db2 = self.delta_output
        self.dW1 = [
            [self.delta_hidden[j] * X[i] for j in range(len(self.delta_hidden))]
            for i in range(len(X))
        ]
        self.db1 = self.delta_hidden

# Compute the Hessian
def compute_hessian(nn, X, y):
    nn.forward(X)
    nn.backward(X, y)

    # Hessian for second layer weights
    H2 = [
        [
            nn.hidden_output[j] * nn.hidden_output[j_prime] * sum(
                nn.delta_output[k] ** 2 for k in range(len(nn.delta_output))
            )
            for j in range(len(nn.hidden_output))
        ]
        for j_prime in range(len(nn.hidden_output))
    ]

    # Hessian for first layer weights
    H1 = [
        [
            X[i] * X[i_prime] * nn.delta_hidden[j] ** 2
            for j in range(len(nn.delta_hidden))
        ]
        for i in range(len(X))
        for i_prime in range(len(X))
    ]

    # Mixed Hessian terms (weight in first layer, weight in second layer)
    H_mix = [
        [
            X[i] * nn.delta_hidden[j] * nn.hidden_output[j] * nn.delta_output[k]
            for k in range(len(nn.delta_output))
        ]
        for i in range(len(X))
        for j in range(len(nn.hidden_output))
    ]

    return H1, H2, H_mix

# Example usage
input_size = 2
hidden_size = 3
output_size = 1
nn = NeuralNetwork(input_size, hidden_size, output_size)

# Input and target output
X = [0.5, -1.2]
y = [0.8]

# Compute Hessian
H1, H2, H_mix = compute_hessian(nn, X, y)

print("Hessian for first layer weights (H1):", H1)
print("Hessian for second layer weights (H2):", H2)
print("Mixed Hessian terms (H_mix):", H_mix)
```

## Fast Multiplication by the Hessian

To efficiently compute the product of the Hessian matrix $ H $ with a vector $ v $, we avoid directly constructing $ H $, which requires $ O(W^2) $ storage and computational cost. Instead, we compute the product $ v^T H $ directly using $ O(W) $ operations by applying the operator $ R\{\cdot\} = v^T \nabla $ to the gradient equations.

---

### Forward Propagation

The forward-propagation equations for a two-layer network are:

$$
a_j = \sum_i w_{ji} x_i
$$

$$
z_j = h(a_j)
$$

$$
y_k = \sum_j w_{kj} z_j
$$

Applying $ R\{\cdot\} $, we derive:

$$
R\{a_j\} = \sum_i v_{ji} x_i
$$

$$
R\{z_j\} = h'(a_j) R\{a_j\}
$$

$$
R\{y_k\} = \sum_j w_{kj} R\{z_j\} + \sum_j v_{kj} z_j
$$

---

### Backward Propagation

The standard backpropagation equations are:

$$
\delta_k = y_k - t_k
$$

$$
\delta_j = h'(a_j) \sum_k w_{kj} \delta_k
$$

Using $ R\{\cdot\} $, we derive:

$$
R\{\delta_k\} = R\{y_k\}
$$

$$
R\{\delta_j\} = h''(a_j) R\{a_j\} \sum_k w_{kj} \delta_k + h'(a_j) \sum_k v_{kj} \delta_k + h'(a_j) \sum_k w_{kj} R\{\delta_k\}
$$

---

### Gradient Computation

The gradient equations for the error function $ E $ are:

$$
\frac{\partial E}{\partial w_{kj}} = \delta_k z_j
$$

$$
\frac{\partial E}{\partial w_{ji}} = \delta_j x_i
$$

Using $ R\{\cdot\} $, the Hessian-vector product components are given by:

$$
R\left(\frac{\partial E}{\partial w_{kj}}\right) = R\{\delta_k\} z_j + \delta_k R\{z_j\}
$$

$$
R\left(\frac{\partial E}{\partial w_{ji}}\right) = x_i R\{\delta_j\}
$$

---

### Implementation Notes

- Additional variables introduced:
  - $ R\{a_j\}, R\{z_j\}, R\{\delta_j\} $ for hidden units.
  - $ R\{\delta_k\}, R\{y_k\} $ for output units.
- For each input pattern, compute the values of these quantities using the equations above.
- Use the final expressions to compute the elements of $ v^T H $.

---

### Special Case: Full Hessian Computation

To compute the full Hessian $ H $, set $ v $ to successive unit vectors of the form:

$$
v = (0, 0, \ldots, 1, \ldots, 0)
$$

Each such vector isolates one column of $ H $, making this method equivalent to the backpropagation-based Hessian evaluation described in Section 5.4.5, with some efficiency trade-offs due to redundant computations.


```python
import math

# Activation function and its derivatives
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def sigmoid_second_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x)) * (1 - 2 * sigmoid(x))

# Network forward pass
def forward_pass(X, weights1, weights2):
    a_hidden = [[sum(weights1[j][i] * x for i, x in enumerate(X)) for j in range(len(weights1))]]
    z_hidden = [[sigmoid(a) for a in a_hidden[0]]]
    a_output = [sum(weights2[k][j] * z for j, z in enumerate(z_hidden[0])) for k in range(len(weights2))]
    y_output = a_output  # Linear activation for the output layer
    return a_hidden[0], z_hidden[0], a_output, y_output

# Compute forward and backward terms for Hessian-vector multiplication
def hessian_vector_product(X, T, weights1, weights2, v1, v2):
    # Forward pass
    a_hidden, z_hidden, a_output, y_output = forward_pass(X, weights1, weights2)
    
    # Gradients
    delta_k = [y - t for y, t in zip(y_output, T)]
    delta_j = [sigmoid_derivative(a_hidden[j]) * sum(weights2[k][j] * delta_k[k] for k in range(len(delta_k))) for j in range(len(a_hidden))]

    # Forward terms (R{} operator)
    R_a_hidden = [sum(v1[j][i] * X[i] for i in range(len(X))) for j in range(len(v1))]
    R_z_hidden = [sigmoid_derivative(a_hidden[j]) * R_a_hidden[j] for j in range(len(a_hidden))]
    R_y_output = [sum(weights2[k][j] * R_z_hidden[j] + v2[k][j] * z_hidden[j] for j in range(len(z_hidden))) for k in range(len(v2))]

    # Backward terms (R{} operator)
    R_delta_k = R_y_output
    R_delta_j = [
        sigmoid_second_derivative(a_hidden[j]) * R_a_hidden[j] * sum(weights2[k][j] * delta_k[k] for k in range(len(delta_k))) +
        sigmoid_derivative(a_hidden[j]) * sum(v2[k][j] * delta_k[k] for k in range(len(delta_k))) +
        sigmoid_derivative(a_hidden[j]) * sum(weights2[k][j] * R_delta_k[k] for k in range(len(delta_k)))
        for j in range(len(a_hidden))
    ]

    # Hessian-vector product components
    R_grad_output = [
        [R_delta_k[k] * z_hidden[j] + delta_k[k] * R_z_hidden[j] for j in range(len(z_hidden))]
        for k in range(len(delta_k))
    ]
    R_grad_input = [
        [X[i] * R_delta_j[j] for i in range(len(X))]
        for j in range(len(delta_j))
    ]

    return R_grad_output, R_grad_input

# Example usage
# Input data
X = [0.5, -1.5]  # Input features
T = [1.0]        # Target output

# Random weights
weights1 = [[0.1, 0.2], [0.3, 0.4]]  # Input to hidden
weights2 = [[0.5, 0.6]]              # Hidden to output

# Vector v (perturbation)
v1 = [[0.01, -0.02], [-0.03, 0.04]]  # Same shape as weights1
v2 = [[0.05, -0.06]]                 # Same shape as weights2

# Compute Hessian-vector product
R_grad_output, R_grad_input = hessian_vector_product(X, T, weights1, weights2, v1, v2)

print("Hessian-vector product (output layer):", R_grad_output)
print("Hessian-vector product (input layer):", R_grad_input)
```

```python
import matplotlib.pyplot as plt

# Define forward pass function
def forward_pass(X, weights1, weights2):
    # Forward pass from input layer to hidden layer
    a_hidden = [0] * len(weights1[0])
    for i in range(len(weights1[0])):
        a_hidden[i] = sum(X[j] * weights1[j][i] for j in range(len(X)))

    # Activation function (sigmoid for hidden layer)
    z_hidden = [1 / (1 + 2.71828 ** (-a)) for a in a_hidden]  # Sigmoid activation

    # Forward pass from hidden layer to output layer
    a_output = [0] * len(weights2[0])  # This should match number of output units
    for i in range(len(weights2[0])):
        a_output[i] = sum(z_hidden[j] * weights2[j][i] for j in range(len(z_hidden)))

    # Output layer activation (linear in this case)
    y_output = a_output

    return a_hidden, z_hidden, a_output, y_output

# Define Hessian-vector product calculation
def hessian_vector_product(X, T, weights1, weights2, v1, v2):
    # Compute the gradient of the output layer
    a_hidden, z_hidden, a_output, y_output = forward_pass(X, weights1, weights2)

    R_grad_output = [0] * len(y_output)
    for k in range(len(y_output)):
        R_grad_output[k] = (y_output[k] - T[k]) * v2[k]

    # Compute the gradient of the hidden layer
    R_grad_input = [0] * len(z_hidden)
    for j in range(len(z_hidden)):
        R_grad_input[j] = z_hidden[j] * v1[j]

    return R_grad_output, R_grad_input

# Function to plot forward pass outputs
def plot_forward_pass(X, T, weights1, weights2):
    a_hidden, z_hidden, a_output, y_output = forward_pass(X, weights1, weights2)

    # Plot the activations of the hidden layer (sigmoid)
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(range(len(z_hidden)), z_hidden, color='blue', alpha=0.6, label='Hidden layer activations')
    plt.xlabel('Hidden units')
    plt.ylabel('Activation values')
    plt.title('Hidden Layer Activation')
    plt.legend()

    # Plot the output layer
    plt.subplot(1, 2, 2)
    plt.bar(range(len(y_output)), y_output, color='green', alpha=0.6, label='Output activations')
    plt.xlabel('Output units')
    plt.ylabel('Activation values')
    plt.title('Output Layer Activation')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Function to plot Hessian-vector product for given weights and perturbation vector
def plot_hessian_vector_product(X, T, weights1, weights2, v1, v2):
    # Compute Hessian-vector product
    R_grad_output, R_grad_input = hessian_vector_product(X, T, weights1, weights2, v1, v2)

    # Visualize Hessian-vector product for output layer
    plt.figure(figsize=(12, 6))

    # Plot output layer Hessian-vector product
    plt.subplot(1, 2, 1)
    plt.bar(range(len(R_grad_output)), R_grad_output, color='red', alpha=0.6)
    plt.xlabel('Hessian-vector components')
    plt.ylabel('Value')
    plt.title('Hessian-Vector Product (Output Layer)')
    
    # Plot input layer Hessian-vector product
    plt.subplot(1, 2, 2)
    plt.bar(range(len(R_grad_input)), R_grad_input, color='orange', alpha=0.6)
    plt.xlabel('Hessian-vector components')
    plt.ylabel('Value')
    plt.title('Hessian-Vector Product (Input Layer)')
    
    plt.tight_layout()
    plt.show()

# Example usage for plotting
X = [0.5, -1.5]  # Input features
T = [1.0]        # Target output

# Random weights
weights1 = [[0.1, 0.2], [0.3, 0.4]]  # Input to hidden (2 input, 2 hidden units)
weights2 = [[0.5], [0.6]]             # Hidden to output (2 hidden, 1 output unit)

# Vector v (perturbation)
v1 = [0.01, -0.02]  # Same shape as weights1 (perturbation for hidden weights)
v2 = [0.05]         # Same shape as weights2 (perturbation for output weights)

# Plot forward pass outputs
plot_forward_pass(X, T, weights1, weights2)

# Plot Hessian-vector product results
plot_hessian_vector_product(X, T, weights1, weights2, v1, v2)
```
