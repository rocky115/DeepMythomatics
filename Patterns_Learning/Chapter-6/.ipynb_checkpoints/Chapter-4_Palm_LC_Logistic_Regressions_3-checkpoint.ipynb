## Naive Bayes and Exponential Family Distributions

###  Discrete Features

Let us consider the case of **discrete feature values** $ x_i $. We will first analyze binary features where $ x_i \in \{0, 1\} $. For $ D $ features, a general distribution would require a table of $ 2^D $ numbers for each class, corresponding to $ 2^D - 1 $ independent parameters due to the summation constraint. This grows exponentially with the number of features, motivating the need for a more restricted representation.

### Naive Bayes Assumption

Under the **naive Bayes assumption**, features are treated as conditionally independent given the class $ C_k $. Thus, the class-conditional distributions take the form:

$$
p(x | C_k) = \prod_{i=1}^D \mu_{ki}^{x_i} (1 - \mu_{ki})^{1 - x_i} \tag{4.81}
$$

Here, $ \mu_{ki} $ represents the probability of feature $ i $ being 1, given class $ C_k $.

### Discriminant Function

Substituting the above into the general discriminant function:

$$
a_k(x) = \sum_{i=1}^D \left[ x_i \ln \mu_{ki} + (1 - x_i) \ln (1 - \mu_{ki}) \right] + \ln p(C_k) \tag{4.82}
$$

This is a **linear function** of the input $ x_i $.

For **binary classification** ($ K = 2 $), the posterior probability can be expressed using the **logistic sigmoid** function. Similar results hold for discrete features with $ M > 2 $ states.

---

##  Exponential Family

For both Gaussian-distributed and discrete inputs, posterior class probabilities are generalized linear models with **logistic sigmoid** ($ K = 2 $) or **softmax** ($ K > 2 $) activation functions.

### Exponential Family Distributions

Consider the exponential family in the form:

$$
p(x | \lambda_k) = h(x) g(\lambda_k) \exp \left( \lambda_k^T u(x) \right) \tag{4.83}
$$

Here, $ h(x) $ and $ g(\lambda_k) $ are known functions, and $ \lambda_k $ is the parameter vector for class $ k $.

### Restricted Exponential Family

Restricting to cases where $ u(x) = x $, and introducing a **scaling parameter** \( s \), we have:

$$
p(x | \lambda_k, s) = \frac{1}{s} h\left( \frac{x}{s} \right) g(\lambda_k) \exp \left( \frac{1}{s} \lambda_k^T x \right) \tag{4.84}
$$

Here, each class has its own parameter vector $ \lambda_k $, but all classes share the same scale parameter $ s $.

### Logistic Regression Interpretation

For the **two-class problem**, substituting this form into the posterior probability equation yields:

$$
a(x) = (\lambda_1 - \lambda_2)^T x + \ln g(\lambda_1) - \ln g(\lambda_2) + \ln p(C_1) - \ln p(C_2) \tag{4.85}
$$

This shows the posterior probability is a **logistic sigmoid** function of a linear discriminant.

### Multi-class Case

For $ K $-class problems, substituting into the softmax formulation yields:

$$
a_k(x) = \lambda_k^T x + \ln g(\lambda_k) + \ln p(C_k) \tag{4.86}
$$

Thus, $ a_k(x) $ remains a linear function of $ x $.

---

### Summary

- **Discrete features** under the naive Bayes assumption lead to linear discriminants.
- **Exponential family distributions** extend this to continuous inputs, maintaining a generalized linear model form for both binary and multi-class classification.


```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, log_loss
import matplotlib.pyplot as plt

# Naive Bayes with Binary Features
class NaiveBayesBinary:
    def __init__(self):
        self.class_probs = None
        self.feature_probs = None

    def fit(self, X, y):
        """Fit Naive Bayes for binary features."""
        self.classes = np.unique(y)
        self.class_probs = {cls: np.mean(y == cls) for cls in self.classes}
        self.feature_probs = {
            cls: np.mean(X[y == cls], axis=0) for cls in self.classes
        }

    def predict_proba(self, X):
        """Predict class probabilities."""
        probs = []
        for x in X:
            class_probs = {}
            for cls in self.classes:
                feature_prob = self.feature_probs[cls]
                class_prob = self.class_probs[cls]
                log_prob = np.sum(x * np.log(feature_prob) + (1 - x) * np.log(1 - feature_prob))
                class_probs[cls] = np.exp(log_prob) * class_prob
            total = sum(class_probs.values())
            probs.append({cls: p / total for cls, p in class_probs.items()})
        return probs

    def predict(self, X):
        """Predict the class label."""
        probas = self.predict_proba(X)
        return [max(p, key=p.get) for p in probas]

# Example usage
if __name__ == "__main__":
    # Simulate binary dataset
    np.random.seed(42)
    X = np.random.randint(0, 2, size=(100, 10))
    y = np.random.randint(0, 2, size=100)

    # Split dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train Naive Bayes
    model = NaiveBayesBinary()
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    # Evaluate
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"Log Loss: {log_loss(y_test, [p[1] for p in y_proba]):.4f}")
```

```python
# Import necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

class NaiveBayesBinary:
    def __init__(self):
        self.class_probs = {}
        self.feature_probs = {}

    def fit(self, X, y):
        """Fit Naive Bayes for binary features."""
        n_samples, n_features = X.shape
        classes = np.unique(y)
        
        # Calculate class probabilities P(Ck)
        for cls in classes:
            self.class_probs[cls] = np.sum(y == cls) / n_samples
            # Conditional feature probabilities P(xi | Ck)
            self.feature_probs[cls] = np.sum(X[y == cls], axis=0) / np.sum(y == cls)
        
    def predict_proba(self, X):
        """Predict class probabilities."""
        n_samples, _ = X.shape
        prob_matrix = np.zeros((n_samples, len(self.class_probs)))
        
        for i, x in enumerate(X):
            class_probs = []
            for cls in self.class_probs.keys():
                # P(Ck) * product(P(x|Ck))
                log_prob = np.sum(x * np.log(self.feature_probs[cls]) + (1 - x) * np.log(1 - self.feature_probs[cls]))
                log_prob += np.log(self.class_probs[cls])
                class_probs.append(np.exp(log_prob))
                
            total = sum(class_probs)
            prob_matrix[i] = [p / total for p in class_probs]
        
        return prob_matrix

    def predict(self, X):
        """Predict the class label."""
        probas = self.predict_proba(X)
        return np.argmax(probas, axis=1)

    def evaluate(self, X, y):
        """Evaluate the model's accuracy."""
        preds = self.predict(X)
        accuracy = np.sum(preds == y) / len(y)
        return accuracy

# Visualization Function
def visualize_decision_boundary(X, y, model):
    """Visualize decision boundaries using seaborn."""
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid = np.c_[xx.ravel(), yy.ravel()]
    
    preds = model.predict(grid.reshape(-1, 2))
    Z = preds.reshape(xx.shape)
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="deep")
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
    plt.title("Naive Bayes Decision Boundary")
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate binary data
    np.random.seed(42)
    X = np.random.randint(0, 2, (100, 2))
    y = np.random.randint(0, 2, 100)
    
    # Initialize and train model
    model = NaiveBayesBinary()
    model.fit(X, y)
    
    # Evaluate model
    print("Accuracy:", model.evaluate(X, y))
    
    # Visualize boundaries
    visualize_decision_boundary(X, y, model)
```


## Probabilistic Discriminative Models

### Introduction

For the two-class classification problem, the posterior probability of class $C_1$ is represented by a logistic sigmoid function:

$$
p(C_1 | \mathbf{x}) = \sigma(a(\mathbf{x})),
$$

where \(a(\mathbf{x})\) is a linear function of \(\mathbf{x}\).

For the multi-class case, the posterior probability of class \(C_k\) is given by the softmax function:

$$
p(C_k | \mathbf{x}) = \frac{\exp(a_k(\mathbf{x}))}{\sum_{j=1}^{K} \exp(a_j(\mathbf{x}))}.
$$

This approach utilizes the functional form of generalized linear models, and parameters are determined through **maximum likelihood estimation**.

---

## Generative vs Discriminative Models

### Generative Models

- **Bayesian inference** involves modeling class-conditional densities \(p(\mathbf{x}|C_k)\) and priors \(p(C_k)\).
- Use **Bayes theorem** to calculate the posterior probabilities:

$$
p(C_k | \mathbf{x}) = \frac{p(\mathbf{x}|C_k)p(C_k)}{p(\mathbf{x})}.
$$

- Can generate synthetic data by sampling from $p(\mathbf{x})$.

### Discriminative Models

- Directly model the posterior probabilities $p(C_k|\mathbf{x})$ by maximizing the conditional likelihood.
- Fewer adaptive parameters compared to generative models.
- May lead to better predictive performance when assumptions about $p(\mathbf{x}|C_k)$ are inaccurate.

---

## Fixed Basis Functions

Instead of working directly with the input $\mathbf{x}$, apply a fixed nonlinear transformation:

$$
\boldsymbol{\phi}(\mathbf{x}) = [\phi_0(\mathbf{x}), \phi_1(\mathbf{x}), \dots, \phi_M(\mathbf{x})]^T,
$$

where \(\phi_0(\mathbf{x}) = 1\) acts as a bias term.

### Decision Boundaries

- **Linear in feature space** $\boldsymbol{\phi}$, but **nonlinear in original space** $\mathbf{x}$.
- For example, consider the logistic regression model:

$$
a(\mathbf{x}) = \mathbf{w}^T \boldsymbol{\phi}(\mathbf{x}),
$$

and the corresponding posterior probability:

$$
p(C_1 | \mathbf{x}) = \sigma(\mathbf{w}^T \boldsymbol{\phi}(\mathbf{x})).
$$

### Example

The right plot shows the feature space $(\phi_1, \phi_2)$, where a **linear decision boundary** corresponds to a **nonlinear decision boundary** in the original space.

---

## Advantages of Fixed Basis Functions

- **Simplifies decision boundary modeling.**
- Useful for problems where the data is not linearly separable in the original space.
- Highlights concepts applicable to more complex models with adaptive basis functions.

### Limitations

- Class overlap in $\mathbf{x}$-space persists and may increase.
- Fixed basis functions may not fully capture the complexity of the data.

---

## Conclusion

Probabilistic discriminative models provide a framework for modeling **posterior probabilities** directly. They leverage fixed nonlinear transformations to simplify decision-making while maintaining flexibility through generalized linear models. These models serve as foundational tools in machine learning, setting the stage for more advanced techniques such as **adaptive basis functions**.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate synthetic dataset
def generate_data():
    np.random.seed(42)
    n_samples = 100
    class_1 = np.random.multivariate_normal(mean=[2, 2], cov=[[1, 0.5], [0.5, 1]], size=n_samples)
    class_2 = np.random.multivariate_normal(mean=[-2, -2], cov=[[1, 0.5], [0.5, 1]], size=n_samples)
    X = np.vstack((class_1, class_2))
    y = np.hstack((np.ones(n_samples), np.zeros(n_samples)))
    return X, y

# Define basis functions
def basis_functions(X):
    phi_1 = np.exp(-np.linalg.norm(X - np.array([1, 1]), axis=1)**2)
    phi_2 = np.exp(-np.linalg.norm(X - np.array([-1, -1]), axis=1)**2)
    phi_0 = np.ones(X.shape[0])
    return np.column_stack((phi_0, phi_1, phi_2))

# Logistic loss function and gradient
def logistic_loss(w, X, y):
    predictions = sigmoid(X @ w)
    return -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

def logistic_gradient(w, X, y):
    predictions = sigmoid(X @ w)
    return (X.T @ (predictions - y)) / y.size

# Gradient descent optimizer
def gradient_descent(X, y, learning_rate=0.1, epochs=1000):
    w = np.zeros(X.shape[1])
    for _ in range(epochs):
        w -= learning_rate * logistic_gradient(w, X, y)
    return w

# Decision boundary visualization
def plot_decision_boundary(X, y, w):
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="deep", alpha=0.7)
    
    # Grid for decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid = np.c_[xx.ravel(), yy.ravel()]
    phi_grid = basis_functions(grid)
    Z = sigmoid(phi_grid @ w).reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Decision Boundary')
    plt.show()

# Main function
def main():
    X, y = generate_data()
    phi_X = basis_functions(X)
    w = gradient_descent(phi_X, y)
    plot_decision_boundary(X, y, w)

# Run the implementation
main()
```

## Logistic Regression Implementation Using Maximum Likelihood
This notebook implements logistic regression for binary classification using maximum likelihood estimation (MLE). The model predicts the posterior probability $ p(C_1|\phi) $ using the logistic sigmoid function.

## Mathematical Background

Logistic regression models the posterior probability as:
$$
p(C_1|\phi) = y(\phi) = \sigma(w^T \phi)
$$

Where $ \sigma $ is the logistic sigmoid function:
$$
\sigma(a) = \frac{1}{1 + e^{-a}}
$$

The cross-entropy error function (negative log-likelihood) for the dataset $ \{\phi_n, t_n\} $ is:
$$
E(w) = -\sum_{n=1}^N \left[ t_n \log(y_n) + (1 - t_n) \log(1 - y_n) \right]
$$

The gradient of the error function:
$$
\nabla E(w) = \sum_{n=1}^N (y_n - t_n) \phi_n
$$

This gradient is used to optimize $ w $ using gradient descent.


```python
# Implementation of Logistic Regression

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cross-entropy loss
def cross_entropy_loss(w, X, y):
    predictions = sigmoid(X @ w)
    return -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

# Gradient of the loss function
def gradient(w, X, y):
    predictions = sigmoid(X @ w)
    return X.T @ (predictions - y) / y.size

# Gradient descent optimizer
def gradient_descent(X, y, learning_rate=0.1, epochs=1000):
    w = np.zeros(X.shape[1])
    for _ in range(epochs):
        w -= learning_rate * gradient(w, X, y)
    return w

# Generate synthetic dataset
def generate_data():
    np.random.seed(42)
    N = 100
    class_1 = np.random.multivariate_normal([2, 2], [[1, 0.5], [0.5, 1]], N)
    class_2 = np.random.multivariate_normal([-2, -2], [[1, 0.5], [0.5, 1]], N)
    X = np.vstack((class_1, class_2))
    y = np.hstack((np.ones(N), np.zeros(N)))
    return X, y

# Add bias term
def add_bias(X):
    return np.hstack((np.ones((X.shape[0], 1)), X))

# Plot decision boundary
def plot_decision_boundary(X, y, w):
    sns.scatterplot(x=X[:, 1], y=X[:, 2], hue=y, palette="deep", alpha=0.7)

    x_min, x_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    x_vals = np.linspace(x_min, x_max, 100)
    y_vals = -(w[0] + w[1] * x_vals) / w[2]

    plt.plot(x_vals, y_vals, color="black")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Logistic Regression Decision Boundary")
    plt.show()

# Main function
def main():
    X, y = generate_data()
    X_bias = add_bias(X)
    w = gradient_descent(X_bias, y)
    plot_decision_boundary(X_bias, y, w)

# Run the logistic regression implementation
main()
```

```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import random
import math

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Manual dot product for vectors
def dot_product(v1, v2):
    return sum(x * y for x, y in zip(v1, v2))

# Predict probability
def predict(features, weights):
    z = dot_product(features, weights)
    return sigmoid(z)

# Cross-entropy loss function
def cross_entropy_loss(features, labels, weights):
    loss = 0
    for x, y in zip(features, labels):
        prediction = predict(x, weights)
        loss -= y * math.log(prediction) + (1 - y) * math.log(1 - prediction)
    return loss / len(labels)

# Gradient computation
def compute_gradient(features, labels, weights):
    gradient = [0] * len(weights)
    for x, y in zip(features, labels):
        prediction = predict(x, weights)
        error = prediction - y
        for j in range(len(weights)):
            gradient[j] += error * x[j]
    return [g / len(labels) for g in gradient]

# Gradient descent optimizer
def gradient_descent(features, labels, learning_rate=0.1, epochs=1000):
    weights = [0] * len(features[0])
    for _ in range(epochs):
        gradient = compute_gradient(features, labels, weights)
        weights = [w - learning_rate * g for w, g in zip(weights, gradient)]
    return weights

# Generate synthetic data
def generate_data():
    random.seed(42)
    data_class1 = [[random.uniform(1, 3), random.uniform(1, 3), 1] for _ in range(50)]
    data_class2 = [[random.uniform(-3, -1), random.uniform(-3, -1), 0] for _ in range(50)]
    data = data_class1 + data_class2
    random.shuffle(data)
    features = [[1] + row[:2] for row in data]  # Adding bias term
    labels = [row[2] for row in data]
    return features, labels

# Plot decision boundary
def plot_decision_boundary(features, labels, weights):
    sns.scatterplot(x=[x[1] for x in features], y=[x[2] for x in features], hue=labels, palette="deep", alpha=0.7)

    x_min, x_max = min(x[1] for x in features) - 1, max(x[1] for x in features) + 1
    x_vals = [i / 100.0 for i in range(int(x_min * 100), int(x_max * 100))]
    y_vals = [-(weights[0] + weights[1] * x) / weights[2] for x in x_vals]

    plt.plot(x_vals, y_vals, color="black")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Logistic Regression Decision Boundary")
    plt.show()

# Main function
def main():
    features, labels = generate_data()
    weights = gradient_descent(features, labels)
    plot_decision_boundary(features, labels, weights)

# Execute main
main()
```

## Iterative Reweighted Least Squares (IRLS)

In the case of linear regression models, the maximum likelihood solution under a Gaussian noise assumption has a closed-form solution due to the quadratic dependence of the log-likelihood function on the parameter vector $ \mathbf{w} $. However, for logistic regression, the nonlinearity of the logistic sigmoid function prevents such a closed-form solution.

### Newton-Raphson Method
The error function for logistic regression is concave and has a unique minimum, which can be efficiently minimized using the **Newton-Raphson iterative optimization scheme**. The update rule is given by:

$$
\mathbf{w}^{\text{new}} = \mathbf{w}^{\text{old}} - \mathbf{H}^{-1} \nabla E(\mathbf{w})
$$

where:
- $ \mathbf{H} $ is the **Hessian matrix**: the matrix of second derivatives of $ E(\mathbf{w}) $.
- $ \nabla E(\mathbf{w}) $ is the gradient of the error function.

### Newton-Raphson for Linear Regression
For linear regression, the gradient and Hessian of the sum-of-squares error function are:

$$
\nabla E(\mathbf{w}) = \sum_{n=1}^N (\mathbf{w}^\top \boldsymbol{\phi}_n - t_n) \boldsymbol{\phi}_n = \mathbf{\Phi}^\top (\mathbf{\Phi} \mathbf{w} - \mathbf{t})
$$

$$
\mathbf{H} = \nabla \nabla E(\mathbf{w}) = \sum_{n=1}^N \boldsymbol{\phi}_n \boldsymbol{\phi}_n^\top = \mathbf{\Phi}^\top \mathbf{\Phi}
$$

Substituting these into the Newton-Raphson update gives:

$$
\mathbf{w}^{\text{new}} = (\mathbf{\Phi}^\top \mathbf{\Phi})^{-1} \mathbf{\Phi}^\top \mathbf{t}
$$

which is the standard **least-squares solution**.

### Newton-Raphson for Logistic Regression
For logistic regression, the error function gradient and Hessian are:

$$
\nabla E(\mathbf{w}) = \sum_{n=1}^N (y_n - t_n) \boldsymbol{\phi}_n = \mathbf{\Phi}^\top (\mathbf{y} - \mathbf{t})
$$

$$
\mathbf{H} = \nabla \nabla E(\mathbf{w}) = \sum_{n=1}^N y_n (1 - y_n) \boldsymbol{\phi}_n \boldsymbol{\phi}_n^\top = \mathbf{\Phi}^\top \mathbf{R} \mathbf{\Phi}
$$

where:
- $ y_n = \sigma(\boldsymbol{\phi}_n^\top \mathbf{w}) $ is the logistic sigmoid function.
- $ \mathbf{R} $ is a diagonal matrix with elements $ R_{nn} = y_n (1 - y_n) $.

The Newton-Raphson update becomes:

$$
\mathbf{w}^{\text{new}} = (\mathbf{\Phi}^\top \mathbf{R} \mathbf{\Phi})^{-1} \mathbf{\Phi}^\top \mathbf{R} \mathbf{z}
$$

where:
$$
\mathbf{z} = \mathbf{\Phi} \mathbf{w}^{\text{old}} - \mathbf{R}^{-1} (\mathbf{y} - \mathbf{t})
$$

### Interpretation of IRLS
This update corresponds to solving a **weighted least-squares problem**, where the weights depend on $ \mathbf{w} $. Since $ \mathbf{R} $ changes at each iteration, this method is called **Iterative Reweighted Least Squares (IRLS)**.

The elements of \( \mathbf{R} \) can be interpreted as variances:
- $ \mathbb{E}[t_n] = y_n $
- $ \text{Var}[t_n] = y_n (1 - y_n) $

The quantity $ z_n $, corresponding to the $ n $-th element of $ \mathbf{z} $, represents an **effective target value** in the linearized space of $ a_n = \boldsymbol{\phi}_n^\top \mathbf{w} $, given by:

$$
z_n = \boldsymbol{\phi}_n^\top \mathbf{w}^{\text{old}} - \frac{y_n - t_n}{y_n (1 - y_n)}
$$


```python
import numpy as np

# Sigmoid function with clipping to avoid overflow
def sigmoid(x):
    # Clip the values to avoid overflow in the exp function
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))

# IRLS for Logistic Regression with Regularization
def irls_logistic_regression(X, y, max_iter=100, tol=1e-6, lambda_reg=1e-4):
    """
    Perform Iterative Reweighted Least Squares (IRLS) for logistic regression with regularization.

    Parameters:
    X : numpy.ndarray
        Design matrix (N x M), where N is the number of samples and M is the number of features.
    y : numpy.ndarray
        Target vector (N x 1), binary labels (0 or 1).
    max_iter : int
        Maximum number of iterations.
    tol : float
        Tolerance for convergence.
    lambda_reg : float
        Regularization parameter.

    Returns:
    w : numpy.ndarray
        Learned weight vector (M x 1).
    """
    
    N, M = X.shape
    # Initialize weights to zeros
    w = np.zeros(M)
    
    for iteration in range(max_iter):
        # Compute the predicted probabilities (sigmoid of the linear model)
        y_pred = sigmoid(X @ w)
        
        # Compute the residuals (difference between predicted and actual labels)
        residual = y_pred - y
        
        # Compute the diagonal matrix R with elements y_n * (1 - y_n)
        R = np.diag(y_pred * (1 - y_pred))
        
        # Compute the gradient and Hessian
        grad = X.T @ residual  # Gradient of the error function
        H = X.T @ R @ X         # Hessian matrix
        
        # Add regularization to the Hessian (Ridge Regularization)
        H += lambda_reg * np.eye(M)
        
        # Update the weights using the Newton-Raphson update rule
        try:
            w_new = w - np.linalg.inv(H) @ grad
        except np.linalg.LinAlgError:
            # If the Hessian is still singular, use the pseudo-inverse
            w_new = w - np.linalg.pinv(H) @ grad
        
        # Check for convergence (if the change in weights is small enough)
        if np.linalg.norm(w_new - w) < tol:
            print(f"Converged in {iteration + 1} iterations.")
            break
        
        # Update the weights
        w = w_new
    
    return w

# Example usage
if __name__ == "__main__":
    # Generate some synthetic data for demonstration
    np.random.seed(42)
    N = 100  # number of samples
    M = 2    # number of features (including intercept)
    
    # Design matrix (with a bias term, i.e., column of ones)
    X = np.hstack([np.ones((N, 1)), np.random.randn(N, M - 1)])
    
    # True weights (for generating labels)
    true_w = np.array([0.5, -0.8])
    
    # Generate synthetic binary target labels using the logistic model
    y = (sigmoid(X @ true_w) > 0.5).astype(int)
    
    # Apply IRLS to learn the weights
    learned_w = irls_logistic_regression(X, y)
    
    print("Learned weights:", learned_w)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import random
import numpy as np

# Sigmoid function to compute probabilities
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate synthetic dataset for binary classification (without using numpy)
random.seed(42)

N = 100  # Number of samples
M = 2    # Number of features (including intercept)

# Design matrix (with a bias term, i.e., column of ones)
X = []
y = []

for i in range(N):
    feature1 = random.gauss(0, 1)  # Generate random data for feature 1
    intercept = 1  # Intercept (bias term)
    X.append([intercept, feature1])
    linear_combination = 0.5 * intercept - 0.8 * feature1  # Using some true weights to create labels
    prob = sigmoid(linear_combination)
    y.append(1 if random.random() < prob else 0)  # Binary outcome based on the probability

# Create a DataFrame for easier plotting with Seaborn
df = pd.DataFrame(X, columns=["Intercept", "Feature 1"])
df['Target'] = y

# 1. Scatter Plot of Data with Logistic Regression Decision Boundary

# Plot data points (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Feature 1", y="Intercept", hue="Target", palette="deep", s=100)

# Create a grid for plotting the decision boundary
xx = np.linspace(min(df["Feature 1"]) - 1, max(df["Feature 1"]) + 1, 100)
yy = np.linspace(min(df["Intercept"]) - 1, max(df["Intercept"]) + 1, 100)
xx, yy = np.meshgrid(xx, yy)

# Compute the decision boundary for each point
Z = np.zeros_like(xx)

# Define the weights for the linear combination
w0, w1 = 0.5, -0.8

# Calculate the logistic function for each point in the grid
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        linear_combination = w0 * xx[i, j] + w1 * yy[i, j]  # Linear model
        Z[i, j] = sigmoid(linear_combination)  # Apply sigmoid to get probabilities

# Plot the decision boundary (contour plot)
plt.contourf(xx, yy, Z, levels=[0, 0.5], cmap="Blues", alpha=0.3)

# Add labels and a legend
plt.title("Logistic Regression Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Intercept")
plt.legend(title="Target", loc="best")
plt.show()

# 2. Plotting Residuals (Predicted Probabilities - Actual Labels)
y_pred = [sigmoid(0.5 * X[i][0] - 0.8 * X[i][1]) for i in range(N)]  # Predicted probabilities
residuals = [y_pred[i] - y[i] for i in range(N)]  # Residuals (difference between predicted and actual)

df_residuals = pd.DataFrame({"Residuals": residuals, "Feature 1": [X[i][1] for i in range(N)], "Target": y})

# Plot residuals using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_residuals, x="Feature 1", y="Residuals", hue="Target", palette="deep")
plt.axhline(0, color='black', linewidth=1, linestyle="--")  # Zero line
plt.title("Logistic Regression Residuals")
plt.xlabel("Feature 1")
plt.ylabel("Residuals")
plt.legend(title="Target", loc="best")
plt.show()

# 3. Plotting the Distribution of Predicted Probabilities
plt.figure(figsize=(8, 6))
sns.histplot(y_pred, kde=True, color="skyblue", bins=20, stat="density")
plt.title("Distribution of Predicted Probabilities")
plt.xlabel("Predicted Probability")
plt.ylabel("Density")
plt.show()

# 4. Confusion Matrix Heatmap
from sklearn.metrics import confusion_matrix

# Predicted class labels (binary)
y_pred_class = [1 if p > 0.5 else 0 for p in y_pred]

# Compute the confusion matrix
cm = confusion_matrix(y, y_pred_class)

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Pred 0", "Pred 1"], yticklabels=["True 0", "True 1"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 5. Correlation Heatmap for multiple features (Using Random Data for More Features)
X_more = []
for i in range(N):
    feature1 = random.gauss(0, 1)
    feature2 = random.gauss(0, 1)
    feature3 = random.gauss(0, 1)
    feature4 = random.gauss(0, 1)
    X_more.append([1, feature1, feature2, feature3, feature4])

df_more = pd.DataFrame(X_more, columns=["Intercept", "Feature 1", "Feature 2", "Feature 3", "Feature 4"])
df_more["Target"] = y

# Compute the correlation matrix
corr_matrix = df_more.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```
