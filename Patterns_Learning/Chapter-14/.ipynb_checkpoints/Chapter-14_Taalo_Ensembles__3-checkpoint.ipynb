## Mixtures of Linear Regression Models using the EM Algorithm

The **Mixture of Linear Regression Models** extends the Gaussian Mixture Model to handle regression tasks. Below, we present the mathematical details, derivations, and the iterative steps of the Expectation-Maximization (EM) algorithm for this model.

## Mixture Model Definition

We model the conditional target distribution $ p(t | \phi, \theta) $ as a mixture of $ K $ linear regression models:
$$
p(t | \phi, \theta) = \sum_{k=1}^K \pi_k \mathcal{N}(t | w_k^T \phi, \beta^{-1}),
$$
where:
- $ \phi $ is the input feature vector,
- $ \pi_k $ are the mixing coefficients $( \sum_k \pi_k = 1 $),
- $ w_k $ are the weight vectors for each component,
- $ \beta $ is the common precision parameter (inverse of variance).

### Parameters
The model parameters $ \theta $ include:
- $ \pi = \{\pi_k\} $: Mixing coefficients,
- $ W = \{w_k\} $: Weight vectors for regression components,
- $ \beta $: Precision parameter.

## EM Algorithm

The EM algorithm alternates between the **E-step** (computing responsibilities) and the **M-step** (updating parameters).

### 1. E-Step: Compute Responsibilities
The responsibilities $ \gamma_{nk} $ are the posterior probabilities for component $ k $ given data point $ n $:
$$
\gamma_{nk} = \frac{\pi_k \mathcal{N}(t_n | w_k^T \phi_n, \beta^{-1})}{\sum_{j=1}^K \pi_j \mathcal{N}(t_n | w_j^T \phi_n, \beta^{-1})}.
$$

### 2. M-Step: Update Parameters

#### Mixing Coefficients $ \pi_k $:
$$
\pi_k = \frac{1}{N} \sum_{n=1}^N \gamma_{nk}.
$$

#### Weights $ w_k $:
Using a weighted least squares formulation:
$$
w_k = \left( \Phi^T R_k \Phi \right)^{-1} \Phi^T R_k t,
$$
where:
- $ R_k = \text{diag}(\gamma_{1k}, \ldots, \gamma_{Nk}) $,
- $ \Phi $ is the design matrix with rows $ \phi_n^T $,
- $ t $ is the vector of target values.

#### Precision $ \beta $:
$$
\beta = \frac{1}{N} \sum_{n=1}^N \sum_{k=1}^K \gamma_{nk} \left( t_n - w_k^T \phi_n \right)^2.
$$

## Example: Synthetic Data

Below, we simulate a dataset and fit a mixture of two linear regression models using the EM algorithm.



````python
#```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
x = np.linspace(-1, 1, 100)
t1 = 0.5 * x + np.random.normal(0, 0.1, len(x))
t2 = -0.5 * x + np.random.normal(0, 0.1, len(x))
x = np.concatenate([x, x])
t = np.concatenate([t1, t2])

# Add bias term
X = np.vstack([x, np.ones_like(x)]).T

# Initialize parameters
K = 2
pi = np.array([0.5, 0.5])
w = [np.random.randn(2) for _ in range(K)]
beta = 1.0

# EM algorithm
num_iterations = 50
N = len(x)
gamma = np.zeros((N, K))

for iteration in range(num_iterations):
    # E-step: Compute responsibilities
    for k in range(K):
        gamma[:, k] = pi[k] * np.exp(-0.5 * beta * (t - X @ w[k])**2)
    gamma /= gamma.sum(axis=1, keepdims=True)

    # M-step: Update parameters
    for k in range(K):
        Nk = gamma[:, k].sum()
        pi[k] = Nk / N
        Rk = np.diag(gamma[:, k])
        w[k] = np.linalg.inv(X.T @ Rk @ X) @ (X.T @ Rk @ t)
    beta = 1 / np.mean([
        gamma[n, k] * (t[n] - X[n] @ w[k])**2
        for n in range(N)
        for k in range(K)
    ])

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(x, t, color="green", label="Data")
colors = ["blue", "red"]
for k in range(K):
    plt.plot(x, X @ w[k], color=colors[k], label=f"Component {k+1}")
plt.title("Mixture of Linear Regression Models")
plt.legend()
plt.show()
````

# Mixtures of Logistic Models

The **Mixture of Logistic Regression Models** extends logistic regression by using a mixture model, allowing for more complex conditional distributions compared to a single logistic regression model. Below, we detail the mathematical framework and the steps of the Expectation-Maximization (EM) algorithm for this model.

---

## Model Definition

The conditional distribution for a target $ t \in \{0, 1\} $, given the feature vector $ \phi $, is defined as:
$$
p(t | \phi, \theta) = \sum_{k=1}^K \pi_k \, y_k^{t} (1 - y_k)^{1-t},
$$
where:
- $ \pi_k $ are the mixing coefficients $( \sum_{k=1}^K \pi_k = 1 $),
- $ y_k = \sigma(w_k^T \phi) $ is the output of the logistic function for component $ k$ ,
- $ \sigma(z) = \frac{1}{1 + e^{-z}} $ is the sigmoid function,
- $ \theta = \{\pi_k, w_k\} $ are the model parameters.

### Data and Likelihood

Given a dataset $ \{(\phi_n, t_n)\}_{n=1}^N $, the likelihood function is:
$$
p(t | \theta) = \prod_{n=1}^N \prod_{k=1}^K \left[ \pi_k \, y_{nk}^{t_n} (1 - y_{nk})^{1-t_n} \right]^{z_{nk}},
$$
where $ z_{nk} $ is a 1-of-$ K $ binary indicator that assigns data point $ n $ to component $ k $. The latent variables $ z_{nk} $ introduce component responsibilities.

---

## EM Algorithm

The EM algorithm alternates between the **E-step** (compute responsibilities) and the **M-step** (update parameters).

### 1. E-Step: Compute Responsibilities

The responsibilities $ \gamma_{nk} $ are the posterior probabilities that data point $ n $ belongs to component $ k $:
$$
\gamma_{nk} = p(z_{nk} = 1 | \phi_n, \theta^{\text{old}}) = \frac{\pi_k \, y_{nk}^{t_n} (1 - y_{nk})^{1-t_n}}{\sum_{j=1}^K \pi_j \, y_{nj}^{t_n} (1 - y_{nj})^{1-t_n}},
$$
where $ y_{nk} = \sigma(w_k^T \phi_n) $.

---

### 2. M-Step: Update Parameters

#### Mixing Coefficients $ \pi_k $:
Maximizing with respect to $ \pi_k $ under the constraint $ \sum_k \pi_k = 1 $ gives:
$$
\pi_k = \frac{1}{N} \sum_{n=1}^N \gamma_{nk}.
$$

#### Weights $ w_k $:
The function to maximize is:
$$
Q(\theta, \theta^{\text{old}}) = \sum_{n=1}^N \sum_{k=1}^K \gamma_{nk} \left[ \ln \pi_k + t_n \ln y_{nk} + (1 - t_n) \ln (1 - y_{nk}) \right].
$$

The gradient with respect to \( w_k \) is:
$$
\nabla_k Q = \sum_{n=1}^N \gamma_{nk} \, (t_n - y_{nk}) \, \phi_n.
$$

The Hessian is:
$$
H_k = -\sum_{n=1}^N \gamma_{nk} \, y_{nk} (1 - y_{nk}) \, \phi_n \phi_n^T.
$$

The weights \( w_k \) are updated iteratively using **Iterative Reweighted Least Squares (IRLS)**:
$$
w_k^{\text{new}} = w_k^{\text{old}} - H_k^{-1} \nabla_k Q.
$$

---

## Example: Mixture of Logistic Models for Classification

Below is an example demonstrating a mixture of logistic regression models applied to a binary classification task.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit  # Sigmoid function

# Generate synthetic data
np.random.seed(42)
x1 = np.random.normal(-2, 1, 50)
x2 = np.random.normal(2, 1, 50)
x = np.concatenate([x1, x2])
t = np.array([0] * 50 + [1] * 50)
phi = np.vstack([x, np.ones_like(x)]).T  # Add bias term

# Initialize parameters
K = 2
pi = np.array([0.5, 0.5])
w = [np.random.randn(2) for _ in range(K)]
gamma = np.zeros((len(x), K))

# EM algorithm
num_iterations = 50
for iteration in range(num_iterations):
    # E-step: Compute responsibilities
    for k in range(K):
        y = expit(phi @ w[k])
        gamma[:, k] = pi[k] * (y**t) * ((1 - y)**(1 - t))
    gamma /= gamma.sum(axis=1, keepdims=True)

    # M-step: Update parameters
    for k in range(K):
        Nk = gamma[:, k].sum()
        pi[k] = Nk / len(x)
        w_grad = np.sum(gamma[:, k][:, None] * (t - expit(phi @ w[k]))[:, None] * phi, axis=0)
        w_hessian = -np.sum(
            gamma[:, k][:, None, None] * expit(phi @ w[k])[:, None] * (1 - expit(phi @ w[k]))[:, None] * 
            np.einsum('ni,nj->nij', phi, phi), axis=0)
        w[k] -= np.linalg.inv(w_hessian) @ w_grad

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(x, t, color="green", label="Data")
colors = ["blue", "red"]
for k in range(K):
    plt.plot(np.sort(x), expit(np.sort(phi @ w[k])), color=colors[k], label=f"Component {k+1}")
plt.title("Mixture of Logistic Regression Models")
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit  # Sigmoid function

# Generate synthetic data
np.random.seed(42)
x1 = np.random.normal(-2, 1, 50)
x2 = np.random.normal(2, 1, 50)
x = np.concatenate([x1, x2])
t = np.array([0] * 50 + [1] * 50)
phi = np.vstack([x, np.ones_like(x)]).T  # Add bias term

# Initialize parameters
K = 2
pi = np.array([0.5, 0.5])
w = [np.random.randn(2) for _ in range(K)]
gamma = np.zeros((len(x), K))

# Regularization term (increase if necessary)
regularization_strength = 1e-4

# EM algorithm
num_iterations = 50
for iteration in range(num_iterations):
    # E-step: Compute responsibilities
    for k in range(K):
        y = expit(phi @ w[k])
        gamma[:, k] = pi[k] * (y**t) * ((1 - y)**(1 - t))
    gamma /= gamma.sum(axis=1, keepdims=True)

    # M-step: Update parameters
    for k in range(K):
        Nk = gamma[:, k].sum()
        pi[k] = Nk / len(x)
        w_grad = np.sum(gamma[:, k][:, None] * (t - expit(phi @ w[k]))[:, None] * phi, axis=0)

        # Correct Hessian computation
        w_hessian = np.zeros((2, 2))  # Hessian is 2x2 because phi is 2-dimensional
        for n in range(len(x)):
            y_n = expit(phi[n] @ w[k])
            w_hessian += gamma[n, k] * y_n * (1 - y_n) * np.outer(phi[n], phi[n])

        # Add a larger regularization term to Hessian
        w_hessian += regularization_strength * np.eye(2)

        # Check if Hessian is invertible (i.e., not singular)
        if np.linalg.det(w_hessian) == 0:
            print(f"Hessian for component {k+1} is singular, adjusting regularization.")
            w_hessian += 1e-3 * np.eye(2)  # Increase the regularization further if singular

        # Use pseudo-inverse instead of direct inversion
        w[k] -= np.linalg.pinv(w_hessian) @ w_grad

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(x, t, color="green", label="Data")
colors = ["blue", "red"]
for k in range(K):
    plt.plot(np.sort(x), expit(np.sort(phi @ w[k])), color=colors[k], label=f"Component {k+1}")
plt.title("Mixture of Logistic Regression Models")
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit  # Sigmoid function

# Generate synthetic data
np.random.seed(42)
x1 = np.random.normal(-2, 1, 50)
x2 = np.random.normal(2, 1, 50)
x = np.concatenate([x1, x2])
t = np.array([0] * 50 + [1] * 50)
phi = np.vstack([x, np.ones_like(x)]).T  # Add bias term

# Normalize the features (important for numerical stability)
phi[:, 0] = (phi[:, 0] - np.mean(phi[:, 0])) / np.std(phi[:, 0])

# Initialize parameters
K = 2  # Number of components in mixture
pi = np.array([0.5, 0.5])  # Mixing coefficients
w = [np.random.randn(2) for _ in range(K)]  # Initialize weights for each component
gamma = np.zeros((len(x), K))  # Responsibility matrix

# Regularization term (to avoid numerical issues)
regularization_strength = 1e-4

# Gradient Descent Settings
learning_rate = 0.01
num_iterations = 50

# EM algorithm with Gradient Descent
for iteration in range(num_iterations):
    # E-step: Compute responsibilities (gamma)
    for k in range(K):
        y = expit(phi @ w[k])  # Sigmoid activation
        gamma[:, k] = pi[k] * (y**t) * ((1 - y)**(1 - t))
    gamma /= gamma.sum(axis=1, keepdims=True)

    # M-step: Update parameters using gradient descent
    for k in range(K):
        # Compute the gradient w.r.t weights (for component k)
        w_grad = np.sum(gamma[:, k][:, None] * (t - expit(phi @ w[k]))[:, None] * phi, axis=0)
        
        # Add regularization to the gradient
        w_grad -= regularization_strength * w[k]

        # Update the weights using gradient descent
        w[k] += learning_rate * w_grad

        # Update mixing coefficients
        Nk = gamma[:, k].sum()
        pi[k] = Nk / len(x)

    # Optionally print the log-likelihood or other metrics
    if iteration % 10 == 0:
        log_likelihood = np.sum(np.log(np.sum(gamma * pi, axis=1)))
        print(f"Iteration {iteration}, Log-Likelihood: {log_likelihood}")

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(x, t, color="green", label="Data")
colors = ["blue", "red"]
for k in range(K):
    plt.plot(np.sort(x), expit(np.sort(phi @ w[k])), color=colors[k], label=f"Component {k+1}")
plt.title("Mixture of Logistic Regression Models")
plt.legend()
plt.show()
````

## Mixture of Experts Model

In Sections 14.5.1 and 14.5.2, we discussed simple mixtures of models such as linear regression and linear classifiers. These models provide a way to extend the flexibility of linear models to represent more complex (e.g., multimodal) predictive distributions. However, they are still limited because the mixing coefficients are typically constants. 

We can increase the capability of such models by allowing the mixing coefficients to be functions of the input variable. This leads to the **Mixture of Experts** (MoE) model, where the mixing coefficients $ \pi_k(x) $ (also known as gating functions) are dependent on the input variable $ x $, and the individual component densities $ p_k(t|x) $ are called experts. 

The general form of the Mixture of Experts model is:

$$
p(t|x) = \sum_{k=1}^{K} \pi_k(x) p_k(t|x)
$$

where:
- $ \pi_k(x) $ is the gating function for expert $ k $,
- $ p_k(t|x) $ is the expert function for component $ k $,
- $ K $ is the number of experts.

## Gating Functions and Constraints



Fig.10 Illustration of a mixture of logistic regression models. The left plot shows data points drawn from two classes denoted red and blue, in which the background colour (which varies from pure red to pure blue) denotes the true probability of the class label. The centre plot shows the result of ﬁtting a single logistic regression model using maximum likelihood, in which the background colour denotes the corresponding probability of the class label. Because the colour is a near-uniform purple, we see that the model assigns a probability of around 0.5 to each of the classes over most of input space. The right plot shows the result of ﬁtting a mixture of two logistic regression models, which now gives much higher probability to the correct labels for many of the points in the blue class.

The gating functions $ \pi_k(x) $ must satisfy the usual constraints for mixing coefficients:
1. $ 0 \leq \pi_k(x) \leq 1 $ for all $ k $,
2. $ \sum_{k=1}^{K} \pi_k(x) = 1 $ for all $ x $.

A common choice for gating functions is the **softmax** function, which ensures these constraints are met. The softmax function for $ K $ classes is given by:

$$
\pi_k(x) = \frac{\exp(z_k(x))}{\sum_{k=1}^{K} \exp(z_k(x))}
$$

where $ z_k(x) $ is the score (logits) for class $ k $, which can be a linear function of the input $ x $.

## Expert Functions

The expert functions $ p_k(t|x) $ can be linear regression or classification models (e.g., logistic regression). Each expert $ k $ can provide a different prediction for the target $ t $, depending on the input $ x $.

## Hierarchical Mixture of Experts (HME)

A more flexible model than the standard Mixture of Experts is the **Hierarchical Mixture of Experts** (HME), where each expert is itself a mixture of experts. The HME introduces a hierarchical structure to the model, allowing for more complex and flexible relationships between the input and output. 

The HME model can be seen as a probabilistic version of decision trees, where the gating functions decide which mixture component is dominant in each region of input space.

## Connection to Mixture Density Networks (MDNs)

The Mixture of Experts model shares similarities with the **Mixture Density Network** (MDN), as both models have multiple components (experts or densities) and use a gating mechanism to decide how to mix these components. However, the key difference lies in the way the models are structured and optimized:

- In the **Mixture of Experts**, each component has a separate gating function, and the components are optimized independently.
- In the **Mixture Density Network**, the gating and expert functions share the hidden units of a neural network, which leads to more complex and non-linear splitting of the input space.

The Mixture of Experts model can be trained efficiently using the **Expectation-Maximization (EM) algorithm**, with **Iterative Reweighted Least Squares (IRLS)** in the M-step.

## Conclusion

The Mixture of Experts model is powerful because it allows different components (experts) to specialize in different regions of the input space, and the gating functions determine which expert is dominant in a given region. This model can be optimized using EM, where each expert and gating model is optimized using convex optimization techniques, though the overall optimization is non-convex.


````python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from scipy.special import softmax

from sklearn.datasets import make_classification

# Create synthetic dataset (binary classification)
# Ensure the sum of informative, redundant, and repeated features is less than the total number of features
X, y = make_classification(
    n_samples=100, 
    n_features=2, 
    n_classes=2, 
    n_informative=2,    # 2 informative features
    n_redundant=0,      # No redundant features
    n_repeated=0,       # No repeated features
    random_state=42
)

# Number of experts
K = 2

# Initialize logistic regression models for each expert
experts = [LogisticRegression() for _ in range(K)]

# Softmax function for gating network
def softmax_function(z):
    return softmax(z, axis=1)

# Fit mixture of experts using EM algorithm
def fit_mixture_of_experts(X, y, K, max_iter=100):
    N, D = X.shape  # N: number of samples, D: number of features
    
    # Initialize weights for gating function (a simple linear model for each expert)
    w_gating = np.random.randn(K, D)
    
    # Initialize the mixture of experts (K logistic regressions)
    for k in range(K):
        experts[k].fit(X, y)  # Fit each expert individually (for initialization)
    
    # EM algorithm
    for _ in range(max_iter):
        # E-step: Compute the responsibility of each expert for each sample
        z = np.dot(X, w_gating.T)  # Linear scores for each expert
        pi_x = softmax_function(z)  # Gating function (responsibility of each expert)

        # M-step: Update experts and gating functions
        # Update experts
        for k in range(K):
            # Weighted logistic regression for expert k (responsibility pi_x[:, k])
            experts[k].fit(X, y, sample_weight=pi_x[:, k])
        
        # Update gating weights using the responsibility pi_x
        for k in range(K):
            # Simple weighted logistic regression for gating function
            w_gating[k] = np.linalg.inv(X.T @ np.diag(pi_x[:, k]) @ X) @ (X.T @ (pi_x[:, k] - 0.5))  # 0.5 as the target for gating function (binary classification)

    return experts, w_gating

# Train the Mixture of Experts
experts, w_gating = fit_mixture_of_experts(X, y, K)
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))  # Subtract max to improve numerical stability
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

# # Predictions from the Mixture of Experts
# def predict_mixture_of_experts(X, experts, w_gating):
#     # Gating function: Compute gating probabilities pi(x)
#     gating_probs = np.softmax(X @ w_gating)  # Assuming `w_gating` is the weight for the gating network
    
#     # Expert predictions
#     expert_preds = np.array([expert.predict(X) for expert in experts]).T  # Transpose so that shape is (100, 2)
    
#     # Combine expert predictions using the gating probabilities
#     final_preds = np.sum(gating_probs * expert_preds, axis=1)  # Sum across the expert dimension (axis 1)
    
#     return final_preds

def predict_mixture_of_experts(X, experts, w_gating):
    # Gating function: Compute gating probabilities pi(x) using softmax
    gating_probs = softmax(X @ w_gating)  # Use custom softmax function
    
    # Expert predictions
    expert_preds = np.array([expert.predict(X) for expert in experts]).T  # Transpose so that shape is (100, 2)
    
    # Combine expert predictions using the gating probabilities
    final_preds = np.sum(gating_probs * expert_preds, axis=1)  # Sum across the expert dimension (axis 1)
    
    return final_preds


# Make predictions
predictions = predict_mixture_of_experts(X, experts, w_gating)

# Evaluate accuracy
accuracy = np.mean((predictions > 0.5) == y)
print(f"Accuracy: {accuracy:.4f}")
````

````python
import random
import math

# Define the softmax function
def softmax(x):
    exp_values = [math.exp(i - max(x)) for i in x]  # Subtract max for numerical stability
    sum_exp_values = sum(exp_values)
    
    # Return the softmax probabilities
    return [i / sum_exp_values for i in exp_values]

# Define a simple expert model (logistic regression model for binary classification)
class LogisticRegressionExpert:
    def __init__(self, weights):
        self.weights = weights
    
    def predict(self, X):
        # Apply sigmoid function to dot product between input and weights for each input in X
        return [self.sigmoid(self.dot_product(x, self.weights)) for x in X]
    
    def sigmoid(self, x):
        return 1 / (1 + math.exp(-x))
    
    def dot_product(self, x, weights):
        return sum(x_i * w_i for x_i, w_i in zip(x, weights))

# Generate a synthetic dataset (for binary classification)
def generate_synthetic_data(n_samples=100, n_features=2, n_classes=2):
    X = []
    y = []
    for _ in range(n_samples):
        x = [random.uniform(-1, 1) for _ in range(n_features)]
        X.append(x)
        y.append(random.choice([0, 1]))  # Random binary class
    return X, y

# Mixture of Experts prediction function
def predict_mixture_of_experts(X, experts, w_gating):
    # Gating function: Compute gating probabilities pi(x) using the custom softmax
    gating_probs = [softmax([sum(x_i * w_g for x_i, w_g in zip(x, w_gating)) for x in X])]
    
    # Expert predictions
    expert_preds = [expert.predict(X) for expert in experts]
    
    # Combine expert predictions using the gating probabilities
    final_preds = []
    for i in range(len(X)):
        final_pred = 0
        for k in range(len(experts)):
            final_pred += gating_probs[0][k] * expert_preds[k][i]
        final_preds.append(final_pred)
    
    return final_preds

# Define the main function for training and evaluation
def train_and_evaluate():
    # Generate synthetic data
    X, y = generate_synthetic_data(n_samples=100, n_features=2)
    
    # Define number of experts
    num_experts = 2
    
    # Initialize experts with random weights (for demonstration purposes)
    experts = [LogisticRegressionExpert(weights=[random.uniform(-1, 1) for _ in range(2)]) for _ in range(num_experts)]
    
    # Define the gating weights (random for demonstration)
    w_gating = [random.uniform(-1, 1) for _ in range(2)]
    
    # Make predictions using the mixture of experts
    predictions = predict_mixture_of_experts(X, experts, w_gating)
    
    # Evaluate accuracy (simple thresholding for binary classification)
    correct_predictions = sum((1 if (pred >= 0.5 and true == 1) or (pred < 0.5 and true == 0) else 0) for pred, true in zip(predictions, y))
    accuracy = correct_predictions / len(y)
    
    return accuracy

# Run the training and evaluation
accuracy = train_and_evaluate()
print(f"Accuracy of the Mixture of Experts model: {accuracy * 100:.2f}%")
````

````python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

# Define the softmax function
def softmax(x):
    exp_values = [math.exp(i - max(x)) for i in x]  # Subtract max for numerical stability
    sum_exp_values = sum(exp_values)
    
    # Return the softmax probabilities
    return [i / sum_exp_values for i in exp_values]

# Define a simple expert model (logistic regression model for binary classification)
class LogisticRegressionExpert:
    def __init__(self, weights):
        self.weights = weights
    
    def predict(self, X):
        # Apply sigmoid function to dot product between input and weights for each input in X
        return [self.sigmoid(self.dot_product(x, self.weights)) for x in X]
    
    def sigmoid(self, x):
        return 1 / (1 + math.exp(-x))
    
    def dot_product(self, x, weights):
        return sum(x_i * w_i for x_i, w_i in zip(x, weights))

# Generate a synthetic dataset (for binary classification)
def generate_synthetic_data(n_samples=100, n_features=2, n_classes=2):
    X = []
    y = []
    for _ in range(n_samples):
        x = [random.uniform(-1, 1) for _ in range(n_features)]
        X.append(x)
        y.append(random.choice([0, 1]))  # Random binary class
    return X, y

# Mixture of Experts prediction function
def predict_mixture_of_experts(X, experts, w_gating):
    # Gating function: Compute gating probabilities pi(x) using the custom softmax
    gating_probs = []
    for x in X:
        gating_probs.append(softmax([sum(x_i * w_g for x_i, w_g in zip(x, w_gating)) for w_gating in w_gating]))
    
    # Expert predictions
    expert_preds = [expert.predict(X) for expert in experts]
    
    # Combine expert predictions using the gating probabilities
    final_preds = []
    for i in range(len(X)):
        final_pred = 0
        for k in range(len(experts)):
            final_pred += gating_probs[i][k] * expert_preds[k][i]
        final_preds.append(final_pred)
    
    return final_preds

# Define the main function for training and evaluation
def train_and_evaluate():
    # Generate synthetic data
    X, y = generate_synthetic_data(n_samples=100, n_features=2)
    
    # Define number of experts
    num_experts = 2
    
    # Initialize experts with random weights (for demonstration purposes)
    experts = [LogisticRegressionExpert(weights=[random.uniform(-1, 1) for _ in range(2)]) for _ in range(num_experts)]
    
    # Define the gating weights (random for demonstration)
    w_gating = [[random.uniform(-1, 1) for _ in range(2)] for _ in range(num_experts)]
    
    # Make predictions using the mixture of experts
    predictions = predict_mixture_of_experts(X, experts, w_gating)
    
    # Evaluate accuracy (simple thresholding for binary classification)
    correct_predictions = sum((1 if (pred >= 0.5 and true == 1) or (pred < 0.5 and true == 0) else 0) for pred, true in zip(predictions, y))
    accuracy = correct_predictions / len(y)
    
    return X, y, experts, predictions, accuracy

# Plotting function
def plot_decision_boundaries(X, y, experts, predictions):
    # Create a grid of points to plot decision boundaries
    x_min, x_max = np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1
    y_min, y_max = np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    
    # Plot each expert's decision boundary
    plt.figure(figsize=(10, 8))
    
    for k, expert in enumerate(experts):
        # Get expert predictions on grid points
        expert_preds = expert.predict(grid_points)
        expert_preds = np.array(expert_preds).reshape(xx.shape)
        
        plt.contourf(xx, yy, expert_preds, levels=[0, 0.5, 1], alpha=0.3, cmap=plt.cm.coolwarm, linestyles="solid")
    
    # Plot the final mixture decision boundary
    final_preds = predict_mixture_of_experts(grid_points, experts, w_gating)
    final_preds = np.array(final_preds).reshape(xx.shape)
    plt.contourf(xx, yy, final_preds, levels=[0, 0.5, 1], alpha=0.5, cmap=plt.cm.coolwarm, linestyles="dashed")
    
    # Plot the true labels
    plt.scatter([x[0] for x in X], [x[1] for x in X], c=y, cmap=plt.cm.coolwarm, edgecolors="k", marker='o')
    plt.title('Mixture of Experts and Expert Decision Boundaries')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.colorbar()
    plt.show()

# Run the training and evaluation
X, y, experts, predictions, accuracy = train_and_evaluate()

# Convert X to a numpy array for plotting purposes
X_np = np.array(X)

# Plot the decision boundaries
plot_decision_boundaries(X_np, y, experts, predictions)

# Print accuracy
print(f"Accuracy of the Mixture of Experts model: {accuracy * 100:.2f}%")
````
