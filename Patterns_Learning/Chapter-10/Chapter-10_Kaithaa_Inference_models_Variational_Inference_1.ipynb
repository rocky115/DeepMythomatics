## Variational Inference and Posterior Approximation

---

## Introduction

A central task in the application of probabilistic models is the evaluation of the posterior distribution $ p(Z \mid X) $ of the latent variables $ Z $ given the observed (visible) data variables $ X $. This also includes the evaluation of expectations computed with respect to this distribution. 

In many practical scenarios:
- The **dimensionality** of the latent space is too high to work with directly.
- The posterior distribution has a **complex form** that makes exact evaluation of expectations infeasible.

---

## Challenges in Posterior Evaluation

1. **Continuous Variables**:
    - Integrations may not have closed-form analytical solutions.
    - The dimensionality of the space and the complexity of the integrand may prohibit numerical integration.

2. **Discrete Variables**:
    - Marginalizations involve summing over all possible configurations of the hidden variables.
    - Exact calculation becomes expensive due to the exponential growth of hidden states.

To address these challenges, approximation schemes are employed. These are broadly categorized into:

- **Stochastic Approximations** (e.g., Markov chain Monte Carlo)
- **Deterministic Approximations** (e.g., Variational Inference)

---

## Variational Inference

Variational inference methods approximate the posterior distribution $ p(Z \mid X) $ by introducing a simpler family of distributions $ q(Z) $ and minimizing the **KL divergence** between the true posterior $ p(Z \mid X) $ and the approximating distribution $ q(Z) $.

### Key Decomposition

$$
\ln p(X) = L(q) + \text{KL}(q \parallel p)
$$

Where:

1. **Lower Bound $ L(q) $**:
   $$
   L(q) = \int q(Z) \ln \frac{p(X, Z)}{q(Z)} \, dZ
   $$

2. **KL Divergence**:
   $$
   \text{KL}(q \parallel p) = - \int q(Z) \ln \frac{p(Z \mid X)}{q(Z)} \, dZ
   $$

By maximizing $ L(q) $, we equivalently minimize $ \text{KL}(q \parallel p) $. The optimization problem is often constrained by selecting a tractable family of distributions $ q(Z) $.

---

## Functional Perspective

Variational methods involve optimizing functionals—mappings that take functions as inputs and return scalar values. An example is **entropy**:

$$
H[p] = - \int p(x) \ln p(x) \, dx
$$

Functional derivatives express how a functional changes in response to infinitesimal variations in the input function.

---

## Optimization in Variational Inference

To restrict $ q(Z) $ to a tractable family:
- Assume a **parametric distribution** $ q(Z \mid \omega) $ governed by parameters $ \omega $.
- Optimize $ L(q) $ with respect to $ \omega $ using nonlinear optimization techniques.

---

## Conclusion

Variational inference provides a powerful framework for approximating complex posterior distributions, balancing flexibility and tractability. Its strengths complement stochastic methods, enabling scalable Bayesian inference in high-dimensional and complex models.


```python
import math
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Define helper functions
def gaussian(x, mu, sigma):
    """Gaussian probability density function."""
    coef = 1 / (math.sqrt(2 * math.pi) * sigma)
    exp = math.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return coef * exp

def log_gaussian(x, mu, sigma):
    """Log of Gaussian probability density function."""
    return -0.5 * math.log(2 * math.pi) - math.log(sigma) - 0.5 * ((x - mu) / sigma) ** 2

def elbo(q_mu, q_sigma, p_mu, p_sigma, n_samples=1000):
    """
    Compute the Evidence Lower Bound (ELBO).
    ELBO = E_q[log p(X, Z)] - E_q[log q(Z)]
    """
    samples = [random.gauss(q_mu, q_sigma) for _ in range(n_samples)]
    log_joint = [log_gaussian(x, p_mu, p_sigma) for x in samples]
    log_q = [log_gaussian(x, q_mu, q_sigma) for x in samples]
    return sum(log_joint) / n_samples - sum(log_q) / n_samples

def optimize_variational(p_mu, p_sigma, init_mu, init_sigma, lr=0.1, n_iter=100):
    """
    Variational optimization using gradient descent.
    """
    q_mu, q_sigma = init_mu, init_sigma
    elbo_values = []

    for _ in range(n_iter):
        # Compute ELBO
        current_elbo = elbo(q_mu, q_sigma, p_mu, p_sigma)
        elbo_values.append(current_elbo)
        
        # Gradients (approximation using samples)
        n_samples = 1000
        samples = [random.gauss(q_mu, q_sigma) for _ in range(n_samples)]
        grad_mu = sum([(x - q_mu) / (q_sigma**2) for x in samples]) / n_samples
        grad_sigma = sum([((x - q_mu)**2 - q_sigma**2) / (q_sigma**3) for x in samples]) / n_samples

        # Update variational parameters
        q_mu += lr * grad_mu
        q_sigma += lr * grad_sigma
    
    return q_mu, q_sigma, elbo_values

# True posterior parameters (assumed Gaussian for simplicity)
p_mu, p_sigma = 2, 1

# Initial variational parameters
init_mu, init_sigma = 0, 2

# Optimize variational parameters
optimized_mu, optimized_sigma, elbo_history = optimize_variational(p_mu, p_sigma, init_mu, init_sigma)

# Results
print(f"Optimized Mean (q_mu): {optimized_mu}")
print(f"Optimized Std Dev (q_sigma): {optimized_sigma}")

# Plot ELBO convergence
plt.figure(figsize=(10, 5))
plt.plot(elbo_history, label="ELBO")
plt.xlabel("Iteration")
plt.ylabel("ELBO")
plt.title("Convergence of ELBO")
plt.legend()
plt.show()

# Visualize true vs. approximated distributions
x_values = [i / 10.0 for i in range(-50, 100)]
true_distribution = [gaussian(x, p_mu, p_sigma) for x in x_values]
variational_distribution = [gaussian(x, optimized_mu, optimized_sigma) for x in x_values]

plt.figure(figsize=(10, 5))
sns.lineplot(x=x_values, y=true_distribution, label="True Posterior (p)")
sns.lineplot(x=x_values, y=variational_distribution, label="Variational Approximation (q)")
plt.title("True vs. Variational Approximation")
plt.xlabel("Z")
plt.ylabel("Density")
plt.legend()
plt.show()
```

This notebook explores factorized distributions in variational inference, based on the general equations for optimizing the Evidence Lower Bound (ELBO) under a mean-field approximation.

---

## Theory

We assume that the posterior distribution $ q(Z) $ factorizes over disjoint groups $ Z_i $, as follows:

$$
q(Z) = \prod_{i=1}^M q_i(Z_i)
$$

Our goal is to maximize the ELBO $ L(q) $ by optimizing each factor $ q_j(Z_j) $ while holding the others fixed. This leads to the update equation:

$$
\ln q_j^*(Z_j) = \mathbb{E}_{i \neq j}[\ln p(X, Z)] + \text{const}
$$

Taking the exponential and normalizing:

$$
q_j^*(Z_j) = \frac{\exp(\mathbb{E}_{i \neq j}[\ln p(X, Z)])}{\int \exp(\mathbb{E}_{i \neq j}[\ln p(X, Z)]) dZ_j}
$$

## Factorized Distributions in Variational Inference

This notebook presents the formulation and implementation of factorized distributions for variational inference, leveraging the mean-field approximation.

---

## Factorization Assumption

We assume that the posterior distribution \( q(Z) \) factorizes over disjoint groups \( Z_i \):

$$
q(Z) = \prod_{i=1}^M q_i(Z_i)
$$

This assumption is commonly used in mean-field theory and simplifies the optimization problem by restricting $ q(Z) $ to this factorized form.

---

## Objective: Maximizing the Lower Bound

The Evidence Lower Bound (ELBO) $ L(q) $ is given by:

$$
L(q) = \mathbb{E}_q[\ln p(X, Z)] - \mathbb{E}_q[\ln q(Z)]
$$

Substituting the factorized form $ q(Z) = \prod_{i=1}^M q_i(Z_i) $:

$$
L(q) = \int q_j(Z_j) \left[ \int \prod_{i \neq j} q_i(Z_i) \ln p(X, Z) \, dZ_i \right] dZ_j - \int q_j(Z_j) \ln q_j(Z_j) \, dZ_j + \text{const}
$$

---

## Update Rule for Each Factor

The optimal update for each factor $ q_j(Z_j) $ is derived by maximizing $ L(q) $ with respect to $ q_j $:

$$
\ln q_j^*(Z_j) = \mathbb{E}_{i \neq j}[\ln p(X, Z)] + \text{const}
$$

Rewriting in normalized form:

$$
q_j^*(Z_j) = \frac{\exp(\mathbb{E}_{i \neq j}[\ln p(X, Z)])}{\int \exp(\mathbb{E}_{i \neq j}[\ln p(X, Z)]) dZ_j}
$$

---

## Iterative Solution

The solution involves iteratively updating each factor $ q_j(Z_j) $ while keeping the others fixed. Convergence is guaranteed because $ L(q) $ is convex with respect to each $ q_j(Z_j) $.

1. Initialize $ q_i(Z_i) $ for all $ i $.
2. Cycle through each $ j $ and update $ q_j(Z_j) $ using:

   $$ 
   \ln q_j^*(Z_j) = \mathbb{E}_{i \neq j}[\ln p(X, Z)] + \text{const}
   $$

3. Repeat until convergence.

---




```python
import numpy as np

# Define example values for the precision matrix and mean vector
# Replace these with actual values from your problem setup
prec_joint = np.array([
    [4.0, 1.0],  # Precision matrix elements
    [1.0, 3.0]
])
mu_joint = np.array([0.5, -0.2])  # Mean vector of the joint distribution

# Diagonal of the precision matrix
prec_diag = np.diag(prec_joint)

# Example initializations for the variational factors
mu_factor = [0.0, 0.0]  # Means of q1(z1) and q2(z2)
var_factor = [1.0, 1.0]  # Variances of q1(z1) and q2(z2)

def update_factor(z_other, mu_joint, prec_joint, prec_diag):
    """
    Update the mean and variance for one factor in the variational approximation.
    Parameters:
        z_other: Value of the other variable.
        mu_joint: Mean of the joint distribution.
        prec_joint: Precision matrix of the joint distribution.
        prec_diag: Diagonal elements of the precision matrix.
    Returns:
        Updated mean and variance for the factor.
    """
    mean_update = mu_joint[0] - prec_joint[0, 1] * (z_other - mu_joint[1]) / prec_diag[0]
    variance_update = 1 / prec_diag[0]
    return mean_update, variance_update

# Update factors iteratively
iterations = 10
for _ in range(iterations):
    # Update q1(z1)
    mu_q1, var_q1 = update_factor(mu_factor[1], mu_joint, prec_joint, prec_diag)
    mu_factor[0], var_factor[0] = mu_q1, var_q1

    # Update q2(z2)
    mu_q2, var_q2 = update_factor(mu_factor[0], mu_joint[::-1], prec_joint[::-1, ::-1], prec_diag[::-1])
    mu_factor[1], var_factor[1] = mu_q2, var_q2

print("Final means:", mu_factor)
print("Final variances:", var_factor)
```

## Factorized Approximations in Variational Inference

The following discusses approximating a Gaussian distribution $ p(z) $ using factorized Gaussian distributions $ q(z) $, analyzing the properties of such approximations.

## Problem Setup

We consider a Gaussian distribution $ p(z) $ over two correlated variables $ z = (z_1, z_2) $, defined as:
$$
p(z) = \mathcal{N}(z \mid \mu, \Lambda^{-1}),
$$
where the mean vector $ \mu $ and precision matrix $ \Lambda $ are given by:
$$
\mu = \begin{bmatrix}
\mu_1 \\
\mu_2
\end{bmatrix}, \quad
\Lambda = \begin{bmatrix}
\Lambda_{11} & \Lambda_{12} \\
\Lambda_{21} & \Lambda_{22}
\end{bmatrix},
$$
with $\Lambda_{21} = \Lambda_{12} $.

### Factorized Gaussian Approximation

We approximate $ p(z) $ using a factorized Gaussian:
$$
q(z) = q_1(z_1)q_2(z_2).
$$

### Derivation of $ q_1^\ast(z_1) $

Using the general result for variational inference:
$$
\ln q_1^\ast(z_1) = \mathbb{E}_{z_2}[\ln p(z)] + \text{const},
$$
we expand $ \ln p(z) $:
$$
\ln q_1^\ast(z_1) = -\frac{1}{2} \left( (z_1 - \mu_1)^2 \Lambda_{11} - 2 z_1 \Lambda_{12} (\mathbb{E}[z_2] - \mu_2) \right) + \text{const}.
$$

Completing the square for $ z_1 $, $ q_1^\ast(z_1) $ is Gaussian:
$$
q_1^\ast(z_1) = \mathcal{N}(z_1 \mid m_1, \Lambda_{11}^{-1}),
$$
where the mean $ m_1 $ is given by:
$$
m_1 = \mu_1 - \Lambda_{11}^{-1} \Lambda_{12} (\mathbb{E}[z_2] - \mu_2).
$$

### Derivation of $ q_2^\ast(z_2) $

By symmetry:
$$
q_2^\ast(z_2) = \mathcal{N}(z_2 \mid m_2, \Lambda_{22}^{-1}),
$$
with:
$$
m_2 = \mu_2 - \Lambda_{22}^{-1} \Lambda_{21} (\mathbb{E}[z_1] - \mu_1).
$$

### Iterative Updates

The expectations $ \mathbb{E}[z_1] $ and $ \mathbb{E}[z_2] $ are:
$$
\mathbb{E}[z_1] = m_1, \quad \mathbb{E}[z_2] = m_2.
$$
These equations are coupled and solved iteratively until convergence.

## Properties of Factorized Approximations

1. **Compactness**: Factorized approximations tend to underestimate variance in directions of high correlation.
2. **Mean Accuracy**: The mean of the factorized approximation matches the true mean.
3. **Directionality**: Variance is aligned with the smallest variance of $ p(z) $, underestimating in orthogonal directions.

---

## Visualization of KL Divergences

To analyze the differences between minimizing $ \text{KL}(q \parallel p) $ and $ \text{KL}(p \parallel q) $:

- **Forward KL $( \text{KL}(q \parallel p) $)**: Produces compact approximations.
- **Reverse KL $( \text{KL}(p \parallel q) $)**: Spreads probability mass over low-probability regions of $ p(z) $.

These behaviors are illustrated in contour plots of the approximations.


## Expectation Propagation and Reverse KL Divergence

In this section, we consider the problem of minimizing the reverse Kullback-Leibler (KL) divergence $ \text{KL}(p \parallel q) $ when $ q(Z) $ is a factorized approximation of the form (10.5). 

---

## KL Divergence for Factorized Approximations

The reverse KL divergence can be written as:

$$
\text{KL}(p \parallel q) = - \int p(Z) \ln \left( \prod_{i=1}^M q_i(Z_i) \right) \, dZ + \text{const}
$$

where the constant term is the entropy of $ p(Z) $, which does not depend on $ q(Z) $. To minimize this divergence, we can optimize each factor $ q_j(Z_j) $ independently. This can be done using Lagrange multipliers, yielding the optimal solution:

$$
q_j^\ast(Z_j) = \frac{p(Z) \, dZ_i}{p(Z_j)}
$$

where the optimal solution for $ q_j(Z_j) $ is the marginal distribution of $ p(Z) $.

This approach does not require iteration because it is a closed-form solution. Applying this to a Gaussian distribution $ p(z) $ over a vector $ z $, we can use the following result from earlier:

$$
\text{KL}(p \parallel q) = - \int q(Z) \ln p(Z) \, dZ
$$

This shows the difference between the two forms of KL divergence and highlights the discrepancies in the approximations of the posterior.

--- 

### Illustration of the Two KL Divergence Minimization Methods

- **KL Divergence $ \text{KL}(q \parallel p) $**: Tends to give an approximation where the mean is correct, but the variance is controlled by the direction of smallest variance of $ p(z) $, underestimating the variance in orthogonal directions.
  
- **Reverse KL Divergence $ \text{KL}(p \parallel q) $**: Gives a more spread out approximation, placing significant probability mass in regions with very low probability.

```python
import numpy as np

# Define the true joint Gaussian distribution p(Z)
def p(Z, mu, Lambda):
    """
    Compute the joint Gaussian distribution p(Z) for a 2D Gaussian.
    :param Z: 2D vector (z1, z2)
    :param mu: Mean vector [mu1, mu2]
    :param Lambda: Precision matrix (inverse of covariance)
    :return: Probability density of p(Z)
    """
    Z = np.array(Z)
    mu = np.array(mu)
    Lambda = np.array(Lambda)
    
    diff = Z - mu
    exponent = -0.5 * np.dot(diff.T, np.dot(Lambda, diff))
    norm_const = np.sqrt(np.linalg.det(Lambda) / (2 * np.pi))
    
    return norm_const * np.exp(exponent)

# Marginal for q1(z1) and q2(z2) given by factorization (without iteration)
def q_factorized(mu, Lambda, E_others, Lambda_inv_others):
    """
    Compute the optimal factorized Gaussian q1(z1) and q2(z2).
    :param mu: Mean of the full distribution [mu1, mu2]
    :param Lambda: Precision matrix of the full distribution
    :param E_others: Expectation of the other component (E[z2] for q1 and E[z1] for q2)
    :param Lambda_inv_others: Inverse of the covariance of the other component
    :return: The updated mean for each component
    """
    # For q1(z1)
    m1 = mu[0] - Lambda_inv_others[0, 1] * (E_others[1] - mu[1])
    
    # For q2(z2)
    m2 = mu[1] - Lambda_inv_others[1, 0] * (E_others[0] - mu[0])
    
    return m1, m2

# Example values for the true distribution p(Z)
mu_true = [0, 0]  # Mean vector [mu1, mu2]
Lambda_true = np.array([[1, 0.5], [0.5, 1]])  # Precision matrix (inverse of covariance)

# Initialize the approximating factorized distribution q(Z)
mu_1 = 0  # Initial guess for q1
mu_2 = 0  # Initial guess for q2
E_1 = 0  # Expectation of z1 in q1
E_2 = 0  # Expectation of z2 in q2
Lambda_inv_1 = np.array([[1, 0]])  # Inverse covariance for q1
Lambda_inv_2 = np.array([[1, 0]])  # Inverse covariance for q2

# Initialize the iteration
iterations = 100
for _ in range(iterations):
    # Update q1(z1) based on q2(z2) and vice versa
    m1, m2 = q_factorized(mu_true, Lambda_true, [E_1, E_2], Lambda_true)
    
    # Update the means
    mu_1 = m1
    mu_2 = m2
    
    # Update expectations (marginals of q1 and q2)
    E_1 = mu_1  # E[z1] = mean of q1
    E_2 = mu_2  # E[z2] = mean of q2
    
    # Optionally, print the updates to observe convergence
    if _ % 10 == 0:
        print(f"Iteration {_}: mu_1 = {mu_1}, mu_2 = {mu_2}")

# Final approximated distribution after optimization
print(f"Final Approximation: mu_1 = {mu_1}, mu_2 = {mu_2}")
```

From regions of $Z$ space in which $p(Z)$ is near zero unless $q(Z)$ is also close to zero. Thus, minimizing this form of KL divergence leads to distributions $q(Z)$ that avoid regions in which $p(Z)$ is small. Conversely, the Kullback-Leibler divergence $KL(p \parallel q)$ is minimized by distributions $q(Z)$ that are nonzero in regions where $p(Z)$ is nonzero.

We can gain further insight into the different behavior of the two KL divergences if we consider approximating a multimodal distribution by a unimodal one, as illustrated in Figure 10.3. In practical applications, the true posterior distribution will often be multimodal, with most of the posterior mass concentrated in some number of relatively small regions of parameter space. These multiple modes may arise through nonidentifiability in the latent space or through complex nonlinear dependence on the parameters.

Both types of multimodality were encountered in Chapter 9 in the context of Gaussian mixtures, where they manifested themselves as multiple maxima in the likelihood function. A variational treatment based on the minimization of $KL(q \parallel p)$ will tend to find one of these modes. By contrast, if we were to minimize $KL(p \parallel q)$, the resulting approximations would average across all of the modes and, in the context of the mixture model, would lead to poor predictive distributions (because the average of two good parameter values is typically itself not a good parameter value).

It is possible to make use of $KL(p \parallel q)$ to define a useful inference procedure, but this requires a rather different approach to the one discussed here, and will be considered in detail when we discuss expectation propagation.

The two forms of Kullback-Leibler divergence are members of the alpha family of divergences (Ali and Silvey, 1966; Amari, 1985; Minka, 2005) defined by:

$$
D_\alpha (p \parallel q) = \int p(x)^{\frac{1 + \alpha}{2}} q(x)^{\frac{1 - \alpha}{2}} \, dx \quad \text{for} \quad -\infty < \alpha < \infty
$$

where $\alpha$ is a continuous parameter. The Kullback-Leibler divergence $KL(p \parallel q)$ corresponds to the limit $\alpha \to 1$, whereas $KL(q \parallel p)$ corresponds to the limit $\alpha \to -1$. For all values of $\alpha$, we have:

$$
D_\alpha (p \parallel q) \geq 0, \quad \text{with equality if, and only if,} \quad p(x) = q(x)
$$

Suppose $p(x)$ is a fixed distribution, and we minimize $D_\alpha (p \parallel q)$ with respect to some set of distributions $q(x)$. Then for $\alpha \to -1$ the divergence is zero-forcing, so that any values of $x$ for which $p(x) = 0$ will have $q(x) = 0$, and typically $q(x)$ will under-estimate the support of $p(x)$ and will tend to seek the mode with the largest mass. Conversely, for $\alpha \to 1$ the divergence is zero-avoiding, so that values of $x$ for which $p(x) > 0$ will have $q(x) > 0$, and typically $q(x)$ will stretch to cover all of $p(x)$, and will over-estimate the support of $p(x)$.

When $\alpha = 0$, we obtain a symmetric divergence that is linearly related to the Hellinger distance given by:

$$
D_H (p \parallel q) = \int \left( \sqrt{p(x)} - \sqrt{q(x)} \right)^2 \, dx
$$

The square root of the Hellinger distance is a valid distance metric.


```python
import numpy as np
import matplotlib.pyplot as plt

# Gaussian Distribution: N(mu, sigma^2)
def gaussian(x, mu, sigma):
    return (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Compute KL Divergence for two Gaussian distributions
def kl_divergence(mu_p, sigma_p, mu_q, sigma_q):
    """
    Computes the Kullback-Leibler divergence KL(p || q) between two Gaussians p(x) and q(x)
    where p = N(mu_p, sigma_p) and q = N(mu_q, sigma_q).
    """
    return np.log(sigma_q / sigma_p) + (sigma_p**2 + (mu_p - mu_q)**2) / (2 * sigma_q**2) - 0.5

# Compute Alpha Divergence for two distributions
def alpha_divergence(p, q, alpha):
    """
    Compute the alpha-divergence D_alpha(p || q) for two probability distributions p(x) and q(x)
    given the parameter alpha.
    """
    return np.sum(p**((1 + alpha) / 2) * q**((1 - alpha) / 2))

# Example Gaussian distributions
mu_p, sigma_p = 0, 1    # Mean and std for p(x)
mu_q, sigma_q = 1, 1.5  # Mean and std for q(x)

# Define the range for x
x = np.linspace(-5, 5, 1000)

# Compute the probability density functions (PDFs) for both p(x) and q(x)
p_x = gaussian(x, mu_p, sigma_p)
q_x = gaussian(x, mu_q, sigma_q)

# Plot the distributions
plt.plot(x, p_x, label=f'p(x) = N({mu_p}, {sigma_p**2})', color='blue')
plt.plot(x, q_x, label=f'q(x) = N({mu_q}, {sigma_q**2})', color='red')
plt.title('Gaussian Distributions p(x) and q(x)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# Calculate KL Divergence (p || q)
kl_value = kl_divergence(mu_p, sigma_p, mu_q, sigma_q)
print(f'KL Divergence between p(x) and q(x): {kl_value}')

# Try Alpha Divergence for different values of alpha
alpha_values = [-1, 0, 1]  # Different values for alpha
for alpha in alpha_values:
    alpha_val = alpha_divergence(p_x, q_x, alpha)
    print(f'Alpha Divergence (α={alpha}) between p(x) and q(x): {alpha_val}')
```

```python
import math

# Gaussian Distribution: N(mu, sigma^2)
def gaussian(x, mu, sigma):
    """
    Compute the value of the Gaussian distribution for a given x, mean (mu), and std dev (sigma).
    """
    coefficient = 1 / math.sqrt(2 * math.pi * sigma ** 2)
    exponent = -0.5 * ((x - mu) ** 2) / (sigma ** 2)
    return coefficient * math.exp(exponent)

# KL Divergence for two Gaussian distributions
def kl_divergence(mu_p, sigma_p, mu_q, sigma_q):
    """
    Compute the KL divergence between two Gaussian distributions p(x) = N(mu_p, sigma_p) and
    q(x) = N(mu_q, sigma_q) using the closed-form expression for KL divergence between Gaussians.
    """
    return math.log(sigma_q / sigma_p) + (sigma_p ** 2 + (mu_p - mu_q) ** 2) / (2 * sigma_q ** 2) - 0.5

# Alpha Divergence for two distributions
def alpha_divergence(x_values, p_values, q_values, alpha):
    """
    Compute the alpha-divergence D_alpha(p || q) using a discrete sum.
    x_values: List of x-values for which to compute p(x) and q(x)
    p_values: List of p(x) values at each x
    q_values: List of q(x) values at each x
    alpha: The alpha parameter for the divergence
    """
    total_sum = 0
    for i in range(len(x_values)):
        total_sum += (p_values[i] ** ((1 + alpha) / 2)) * (q_values[i] ** ((1 - alpha) / 2))
    return total_sum * (x_values[1] - x_values[0])  # Approximate the integral by multiplying with delta_x

# Example: Define two Gaussian distributions
mu_p, sigma_p = 0, 1    # Mean and std dev for p(x)
mu_q, sigma_q = 1, 1.5  # Mean and std dev for q(x)

# Define the range for x (discretize it)
x_values = [i * 0.01 for i in range(-500, 501)]  # Discretized range from -5 to 5 with step 0.01

# Compute the values of p(x) and q(x) at each x
p_values = [gaussian(x, mu_p, sigma_p) for x in x_values]
q_values = [gaussian(x, mu_q, sigma_q) for x in x_values]

# Calculate KL Divergence
kl_value = kl_divergence(mu_p, sigma_p, mu_q, sigma_q)
print(f'KL Divergence between p(x) and q(x): {kl_value}')

# Calculate Alpha Divergence for different values of alpha
alpha_values = [-1, 0, 1]  # Different values for alpha
for alpha in alpha_values:
    alpha_val = alpha_divergence(x_values, p_values, q_values, alpha)
    print(f'Alpha Divergence (α={alpha}) between p(x) and q(x): {alpha_val}')
```
