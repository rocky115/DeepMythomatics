## Marginal and Conditional Gaussian Distributions

In this section, we discuss the marginal and conditional distributions of a multivariate Gaussian distribution.

### 1. Joint Distribution

Given a joint Gaussian distribution $ p(\mathbf{x}_a, \mathbf{x}_b) $, the conditional distribution $ p(\mathbf{x}_a | \mathbf{x}_b) $ and the marginal distribution $ p(\mathbf{x}_a) $ can be derived efficiently using the quadratic form in the exponent of the joint distribution. 

The joint distribution can be written as:

$$
p(\mathbf{x}_a, \mathbf{x}_b) = \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}} \exp \left( -\frac{1}{2} \mathbf{x}^T \Lambda \mathbf{x} \right)
$$

where $ \Lambda = \Sigma^{-1} $ is the precision matrix and $ \Sigma $ is the covariance matrix of the joint distribution.

### 2. Conditional Distribution

The conditional distribution $ p(\mathbf{x}_a | \mathbf{x}_b) $ for a Gaussian joint distribution is also Gaussian. The conditional mean $ \mu_{a|b} $ and covariance $ \Sigma_{a|b} $ are derived from the partitioned precision matrix $ \Lambda $. The formula for the conditional distribution is:

$$
p(\mathbf{x}_a | \mathbf{x}_b) = \mathcal{N}(\mathbf{x}_a | \mu_{a|b}, \Sigma_{a|b})
$$

Where:

- **Conditional Mean:**

$$
\mu_{a|b} = \mu_a - \Lambda_{aa}^{-1} \Lambda_{ab} (\mathbf{x}_b - \mu_b)
$$

- **Conditional Covariance:**

$$
\Sigma_{a|b} = \Lambda_{aa}^{-1}
$$

### 3. Marginal Distribution

The marginal distribution $ p(\mathbf{x}_a) $ is obtained by integrating out $ \mathbf{x}_b $ from the joint distribution:

$$
p(\mathbf{x}_a) = \int p(\mathbf{x}_a, \mathbf{x}_b) d\mathbf{x}_b
$$

By completing the square in the quadratic form, we can find the mean and covariance of the marginal distribution:

- **Marginal Mean:**

$$
\mu_a = \mu_a
$$

- **Marginal Covariance:**

$$
\Sigma_a = \Sigma_{aa} - \Sigma_{ab} \Sigma_{bb}^{-1} \Sigma_{ba}
$$

### 4. Partitioned Gaussian

Given a joint Gaussian distribution with the partitioned mean vector and covariance matrix:

$$
\mathbf{x} =
\begin{pmatrix}
\mathbf{x}_a \\
\mathbf{x}_b
\end{pmatrix}, \quad
\mu = 
\begin{pmatrix}
\mu_a \\
\mu_b
\end{pmatrix}, \quad
\Sigma =
\begin{pmatrix}
\Sigma_{aa} & \Sigma_{ab} \\
\Sigma_{ba} & \Sigma_{bb}
\end{pmatrix}
$$

The conditional and marginal distributions are:

- **Conditional Distribution:**

$$
p(\mathbf{x}_a | \mathbf{x}_b) = \mathcal{N}(\mathbf{x}_a | \mu_{a|b}, \Sigma_{a|b})
$$

$$
\mu_{a|b} = \mu_a - \Sigma_{aa}^{-1} \Sigma_{ab} (\mathbf{x}_b - \mu_b)
$$

- **Marginal Distribution:**

$$
p(\mathbf{x}_a) = \mathcal{N}(\mathbf{x}_a | \mu_a, \Sigma_{aa})
$$

### Conclusion

For a joint Gaussian distribution, the marginal and conditional distributions are also Gaussian. The key difference is that the conditional distribution is derived from the precision matrix, while the marginal distribution is derived from the covariance matrix. This illustrates the important relationships between different forms of Gaussian distributions.



```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute conditional Gaussian distribution
def conditional_gaussian(xb, mu_a, mu_b, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb):
    """
    Compute the conditional mean and covariance of the Gaussian distribution.
    
    Arguments:
    xb : numpy array
        Value of xb (conditional variable)
    mu_a, mu_b : numpy arrays
        Mean vectors for xa and xb
    Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb : numpy arrays
        Covariance matrices for xa, xb, and cross terms
    
    Returns:
    mu_a_b : numpy array
        Conditional mean of xa given xb
    Sigma_a_b_inv : numpy array
        Inverse conditional covariance of xa given xb
    """
    Lambda_ab = np.linalg.inv(Sigma_bb)
    xb_diff = xb - mu_b
    Lambda_ab_inv = np.dot(Lambda_ab, xb_diff)
    
    # Compute the conditional mean
    mu_a_b = mu_a + np.dot(Sigma_ab.T, Lambda_ab_inv)
    
    # Compute the conditional covariance
    Sigma_a_b_inv = np.linalg.inv(Sigma_aa - np.dot(np.dot(Sigma_ab.T, Lambda_ab), Sigma_ab))
    
    return mu_a_b, Sigma_a_b_inv

# Function to compute marginal Gaussian distribution
def marginal_gaussian(mu_a, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb):
    """
    Compute the marginal distribution of xa from the joint Gaussian.
    
    Arguments:
    mu_a : numpy array
        Mean vector for xa
    Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb : numpy arrays
        Covariance matrices for xa, xb, and cross terms
    
    Returns:
    mu_a : numpy array
        Mean of the marginal distribution of xa
    Sigma_a : numpy array
        Covariance of the marginal distribution of xa
    """
    # Marginal covariance is simply:
    Sigma_a = Sigma_aa - np.dot(np.dot(Sigma_ab.T, np.linalg.inv(Sigma_bb)), Sigma_ba)
    
    return mu_a, Sigma_a

# Example of Gaussian distributions
def plot_gaussian_distributions():
    # Define the means
    mu_a = np.array([0])
    mu_b = np.array([0])
    
    # Define the covariance matrices
    Sigma_aa = np.array([[1]])
    Sigma_ab = np.array([[0.5]])
    Sigma_ba = np.array([[0.5]])
    Sigma_bb = np.array([[1]])
    
    # Conditional distribution for xb = 0.7
    xb_value = np.array([0.7])
    mu_a_b, Sigma_a_b_inv = conditional_gaussian(xb_value, mu_a, mu_b, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb)
    
    # Marginal distribution of xa
    mu_a_marginal, Sigma_a_marginal = marginal_gaussian(mu_a, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb)
    
    # Plot the marginal and conditional distributions
    xa_values = np.linspace(-3, 3, 100)
    marginal_dist = (1 / (2 * np.pi * np.linalg.det(Sigma_a_marginal)) ** 0.5) * \
                    np.exp(-0.5 * (xa_values - mu_a_marginal) ** 2 / Sigma_a_marginal)
    
    conditional_dist = (1 / (2 * np.pi * np.linalg.det(Sigma_a_b_inv)) ** 0.5) * \
                       np.exp(-0.5 * (xa_values - mu_a_b) ** 2 / Sigma_a_b_inv)
    
    plt.figure(figsize=(10, 6))
    plt.plot(xa_values, marginal_dist, label='Marginal Distribution $p(x_a)$', color='blue')
    plt.plot(xa_values, conditional_dist, label='Conditional Distribution $p(x_a | x_b = 0.7)$', color='red')
    plt.title("Marginal and Conditional Gaussian Distributions")
    plt.xlabel("$x_a$")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the plotting function to visualize the distributions
plot_gaussian_distributions()
```

```python
import math
import matplotlib.pyplot as plt

# Function to compute the inverse of a 1x1 matrix
def inverse_1x1(matrix):
    # Inverse of a 1x1 matrix is just the reciprocal of its single element
    if matrix[0][0] == 0:
        raise ValueError("Matrix is singular and cannot be inverted.")
    return [[1 / matrix[0][0]]]

# Function to multiply a 1x1 matrix by a vector
def mat_vec_multiply(matrix, vector):
    # Since it's 1x1, this is just multiplication by a scalar
    return [matrix[0][0] * vector[0]]

# Function to compute the conditional Gaussian distribution
def conditional_gaussian(xb, mu_a, mu_b, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb):
    """
    Compute the conditional mean and covariance of the Gaussian distribution.
    
    Arguments:
    xb : list
        Value of xb (conditional variable)
    mu_a, mu_b : lists
        Mean vectors for xa and xb
    Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb : lists
        Covariance matrices for xa, xb, and cross terms
    
    Returns:
    mu_a_b : list
        Conditional mean of xa given xb
    Sigma_a_b_inv : list
        Inverse conditional covariance of xa given xb
    """
    Lambda_ab = inverse_1x1(Sigma_bb)  # Inverse of Sigma_bb (1x1 matrix)
    xb_diff = [xb[0] - mu_b[0]]
    
    Lambda_ab_inv = mat_vec_multiply(Lambda_ab, xb_diff)
    
    # Compute the conditional mean
    mu_a_b = [mu_a[0] + Lambda_ab_inv[0] * Sigma_ab[0][0]]
    
    # Compute the conditional covariance (for 1x1 matrix, it's just Sigma_aa)
    Sigma_a_b_inv = Lambda_ab  # Since it's 1x1, conditional covariance is the inverse of Sigma_bb
    
    return mu_a_b, Sigma_a_b_inv

# Function to compute the marginal Gaussian distribution
def marginal_gaussian(mu_a, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb):
    """
    Compute the marginal distribution of xa from the joint Gaussian.
    
    Arguments:
    mu_a : list
        Mean vector for xa
    Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb : lists
        Covariance matrices for xa, xb, and cross terms
    
    Returns:
    mu_a : list
        Mean of the marginal distribution of xa
    Sigma_a : list
        Covariance of the marginal distribution of xa
    """
    Sigma_ba_inv = inverse_1x1(Sigma_bb)  # Inverse of Sigma_bb (1x1 matrix)
    Sigma_a = [[Sigma_aa[0][0] - Sigma_ab[0][0] * Sigma_ba_inv[0][0] * Sigma_ba[0][0]]]
    
    return mu_a, Sigma_a

# Function to plot Gaussian distributions
def plot_gaussian_distributions():
    # Define the means (1x1 case)
    mu_a = [0]
    mu_b = [0]
    
    # Define the covariance matrices (1x1 case)
    Sigma_aa = [[1]]
    Sigma_ab = [[0.5]]
    Sigma_ba = [[0.5]]
    Sigma_bb = [[1]]
    
    # Conditional distribution for xb = 0.7
    xb_value = [0.7]
    mu_a_b, Sigma_a_b_inv = conditional_gaussian(xb_value, mu_a, mu_b, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb)
    
    # Marginal distribution of xa
    mu_a_marginal, Sigma_a_marginal = marginal_gaussian(mu_a, Sigma_aa, Sigma_ab, Sigma_ba, Sigma_bb)
    
    # Plot the marginal and conditional distributions
    xa_values = [i * 0.1 for i in range(-30, 31)]  # Range from -3 to 3
    
    def gaussian(x, mu, sigma):
        return (1 / (math.sqrt(2 * math.pi * sigma))) * math.exp(-0.5 * (x - mu) ** 2 / sigma)
    
    marginal_dist = [gaussian(xa, mu_a_marginal[0], Sigma_a_marginal[0][0]) for xa in xa_values]
    
    conditional_dist = [gaussian(xa, mu_a_b[0], Sigma_a_b_inv[0][0]) for xa in xa_values]
    
    plt.figure(figsize=(10, 6))
    plt.plot(xa_values, marginal_dist, label='Marginal Distribution $p(x_a)$', color='blue')
    plt.plot(xa_values, conditional_dist, label='Conditional Distribution $p(x_a | x_b = 0.7)$', color='red')
    plt.title("Marginal and Conditional Gaussian Distributions")
    plt.xlabel("$x_a$")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the plotting function to visualize the distributions
plot_gaussian_distributions()
```

### 2.3.3 Bayes Theorem for Gaussian Variables

 We have considered a Gaussian $ p(x) $ in which we partitioned the vector $ x $ into two subvectors $ x = (x_a , x_b) $ and then found expressions for the conditional distribution $ p(x_a | x_b) $ and the marginal distribution $ p(x_a) $. We noted that the mean of the conditional distribution $ p(x_a | x_b) $ was a linear function of $ x_b $. Here we shall suppose that we are given a Gaussian marginal distribution $ p(x) $ and a Gaussian conditional distribution $ p(y|x) $ in which $ p(y|x) $ has a mean that is a linear function of $ x $, and a covariance which is independent of $ x $. This is an example of a linear Gaussian model (Roweis and Ghahramani, 1999), which we shall study in greater generality in Section 8.1.4.

We wish to find the marginal distribution $ p(y) $ and the conditional distribution $ p(x|y) $. This is a problem that will arise frequently in subsequent chapters, and it will prove convenient to derive the general results here. We shall take the marginal and conditional distributions to be:

$$
p(x) = \mathcal{N}(x|\mu, \Lambda^{-1}) \tag{2.99}
$$

$$
p(y|x) = \mathcal{N}(y|Ax + b, L^{-1}) \tag{2.100}
$$

where $ \mu $, $ A $, and $ b $ are parameters governing the means, and $ \Lambda $ and $ L $ are precision matrices. If $ x $ has dimensionality $ M $ and $ y $ has dimensionality $ D $, then the matrix $ A $ has size $ D \times M $.

#### 1. Joint Distribution

First, we find an expression for the joint distribution over $ x $ and $ y $. To do this, we define

$$
z = \begin{pmatrix} x \\ y \end{pmatrix} \tag{2.101}
$$

and then consider the log of the joint distribution:

$$
\ln p(z) = \ln p(x) + \ln p(y|x) 
$$

$$
= -\frac{1}{2} (x - \mu)^T \Lambda (x - \mu) - \frac{1}{2} (y - Ax - b)^T L (y - Ax - b) + \text{const} \tag{2.102}
$$

where "const" denotes terms independent of $ x $ and $ y $. As before, we see that this is a quadratic function of the components of $ z $, and hence $p(z) $ is a Gaussian distribution. To find the precision of this Gaussian, we consider the second order terms in (2.102), which can be written as:

$$
= -\frac{1}{2} \begin{pmatrix} x & y \end{pmatrix}^T \begin{pmatrix} \Lambda + A^T L A & -A^T L \\ -L A & L \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} \tag{2.103}
$$

Thus, the Gaussian distribution over $ z $ has precision (inverse covariance) matrix given by:

$$
R = \begin{pmatrix} \Lambda + A^T L A & -A^T L \\ -L A & L \end{pmatrix} \tag{2.104}
$$

The covariance matrix is found by taking the inverse of the precision, which can be done using the matrix inversion formula (2.76) to give:

$$
\text{cov}[z] = R^{-1} = \begin{pmatrix} \Lambda^{-1} + A^T \Lambda^{-1} A & A^T \Lambda^{-1} \\ \Lambda^{-1} A & L^{-1} \end{pmatrix} \tag{2.105}
$$

Similarly, we can find the mean of the Gaussian distribution over $ z $ by identifying the linear terms in (2.102), which are given by:

$$
\begin{pmatrix} \Lambda \mu - A^T L b \\ L b \end{pmatrix} \tag{2.106}
$$

Using our earlier result (2.71) obtained by completing the square over the quadratic form of a multivariate Gaussian, we find that the mean of $ z $ is given by:

$$
E[z] = \begin{pmatrix} \Lambda \mu - A^T L b \\ L b \end{pmatrix} R^{-1} \tag{2.107}
$$

From (2.105), we then obtain:

$$
E[z] = \begin{pmatrix} A \mu + b \end{pmatrix} \tag{2.108}
$$

#### 2. Marginal Distribution of $ y $

Next, we find an expression for the marginal distribution $ p(y) $ in which we have marginalized over $ x $. Recall that the marginal distribution over a subset of the components of a Gaussian random vector takes a particularly simple form when expressed in terms of the partitioned covariance matrix. Specifically, its mean and covariance are given by (2.92) and (2.93), respectively. Making use of (2.105) and (2.108), we see that the mean and covariance of the marginal distribution $ p(y) $ are given by:

$$
E[y] = A \mu + b \tag{2.109}
$$

$$
\text{cov}[y] = L^{-1} + A \Lambda^{-1} A^T \tag{2.110}
$$

A special case of this result is when $ A = I $, in which case it reduces to the convolution of two Gaussians, for which we see that the mean of the convolution is the sum of the mean of the two Gaussians, and the covariance of the convolution is the sum of their covariances.

#### 3. Conditional Distribution of $ x $ Given $ y $

Finally, we seek an expression for the conditional $ p(x|y) $. Recall that the results for the conditional distribution are most easily expressed in terms of the partitioned precision matrix, using (2.73) and (2.75). Applying these results to (2.105) and (2.108), we see that the conditional distribution $ p(x|y) $ has mean and covariance given by:

$$
E[x|y] = (\Lambda + A^T L A)^{-1} A^T L (y - b) + \Lambda \mu \tag{2.111}
$$

$$
\text{cov}[x|y] = (\Lambda + A^T L A)^{-1} \tag{2.112}
$$

The evaluation of this conditional can be seen as an example of Bayes theorem. We can interpret the distribution $ p(x) $ as a prior distribution over $ x $. If the variable $ y $ is observed, then the conditional distribution $ p(x|y) $ represents the corresponding posterior distribution over $ x $.

#### 4. Summary

Given a marginal Gaussian distribution for $ x $ and a conditional Gaussian distribution for $ y $ given $ x $ in the form:

$$
p(x) = \mathcal{N}(x|\mu, \Lambda^{-1}) \tag{2.113}
$$

$$
p(y|x) = \mathcal{N}(y|Ax + b, L^{-1}) \tag{2.114}
$$

the marginal distribution of $ y $ and the conditional distribution of $ x $ given $ y $ are given by:

$$
p(y) = \mathcal{N}(y|A\mu + b, L^{-1} + A \Lambda^{-1} A^T) \tag{2.115}
$$

$$
p(x|y) = \mathcal{N}(x|\Sigma (A^T L (y - b) + \Lambda \mu), \Sigma) \tag{2.116}
$$

where

$$
\Sigma = (\Lambda + A^T L A)^{-1} \tag{2.117}
$$


# Maximum Likelihood Estimation for Gaussian Distribution

Given a dataset \( X = (x_1, x_2, \dots, x_N)^T \), where the observations \( \{ x_n \} \) are assumed to be drawn independently from a multivariate Gaussian distribution, the goal is to estimate the parameters of the Gaussian distribution (mean \( \mu \) and covariance \( \Sigma \)) using Maximum Likelihood Estimation (MLE).

## 2.3.4 Maximum Likelihood for the Gaussian

The log-likelihood function for the Gaussian distribution is given by:

$$
\ln p(X|\mu, \Sigma) = -\frac{N}{2} \ln(2\pi) - \frac{1}{2} \ln|\Sigma| - \frac{1}{2} \sum_{n=1}^N (x_n - \mu)^T \Sigma^{-1} (x_n - \mu)
$$

### Sufficient Statistics

By rearranging the log-likelihood, we observe that the likelihood depends on the dataset only through two quantities:

$$
\sum_{n=1}^N x_n, \quad \sum_{n=1}^N x_n x_n^T
$$

These are the **sufficient statistics** for the Gaussian distribution.

### Estimating the Mean

To find the Maximum Likelihood Estimate (MLE) of the mean \( \mu \), we compute the derivative of the log-likelihood with respect to \( \mu \):

$$
\frac{\partial}{\partial \mu} \ln p(X|\mu, \Sigma) = \Sigma^{-1} \sum_{n=1}^N (x_n - \mu)
$$

Setting this derivative to zero, we obtain the solution for the MLE of the mean:

$$
\mu_{ML} = \frac{1}{N} \sum_{n=1}^N x_n
$$

This is simply the **sample mean** of the observed data points.

### Estimating the Covariance

Maximizing the log-likelihood with respect to \( \Sigma \) involves a more complicated derivation. Ignoring symmetry constraints initially, the solution is derived as:

$$
\Sigma_{ML} = \frac{1}{N} \sum_{n=1}^N (x_n - \mu_{ML})(x_n - \mu_{ML})^T
$$

Note that this estimate for the covariance is **biased**. The expectation of \( \Sigma_{ML} \) is:

$$
E[\Sigma_{ML}] = \Sigma
$$

To correct for this bias, we define the **biased-corrected covariance estimator** \( \Sigma' \):

$$
\Sigma' = \frac{1}{N-1} \sum_{n=1}^N (x_n - \mu_{ML})(x_n - \mu_{ML})^T
$$

This estimator has an expectation equal to the true covariance \( \Sigma \):

$$
E[\Sigma'] = \Sigma
$$

### Expectation of MLE Estimates

Under the true distribution, the expectation of the MLE for the mean and covariance are:

$$
E[\mu_{ML}] = \mu \quad \text{(true mean)}
$$

$$
E[\Sigma_{ML}] = \Sigma \quad \text{(true covariance)}
$$

### Sequential Estimation

Sequential estimation is useful when data points arrive one at a time, and we wish to update our estimates incrementally. Starting with an initial estimate of the mean $ \mu_{ML} $ based on $ N-1 $ data points, the update after observing the $ N $-th data point $ x_N $ is:

$$
\mu_{ML}^{(N)} = \mu_{ML}^{(N-1)} + \frac{1}{N} (x_N - \mu_{ML}^{(N-1)})
$$

This formula gives the same result as the batch MLE estimate as $ N \to \infty $, but it allows for incremental updates as new data points arrive.

## Sequential Learning and the Robbins-Monro Algorithm

Sequential learning can be formalized using the **Robbins-Monro algorithm**, which is designed to find the root of a function $ f(\theta) $. Suppose we have a pair of random variables $ \theta $ and $ z $ governed by a joint distribution $ p(z, \theta) $. The goal is to find the root $ \theta^* $ such that:

$$
f(\theta^*) = E[z|\theta^*] = 0
$$

In sequential learning, we observe $ z $ one data point at a time and update our estimate of $ \theta $ based on each observation. This allows us to estimate the root of $ f(\theta) $ without needing to process all data points at once, making it suitable for online learning and large datasets.

### General Sequential Update Formula

The general sequential update rule for the Robbins-Monro algorithm is:

$$
\theta_{k+1} = \theta_k + \alpha_k (z_k - f(\theta_k))
$$

where:
- $ \theta_k $ is the current estimate of $ \theta $,
- $ \alpha_k $ is the step size (learning rate),
- $ z_k $ is the observation at time $ k $,
- $ f(\theta_k) $ is the current estimate of the regression function.

This update rule allows the estimation of $ \theta $ based on sequential observations.

## Conclusion

We have discussed the process of maximum likelihood estimation for a Gaussian distribution, including the estimation of the mean and covariance, and the use of sequential estimation methods. The Robbins-Monro algorithm provides a general framework for sequential learning, useful for real-time applications and large datasets.


```python
# Define the functions to compute the MLE for Gaussian distribution

def compute_mean_mle(X):
    """
    Compute the Maximum Likelihood Estimate for the mean of a Gaussian distribution.
    
    Parameters:
        X (list of list): List of N data points, each being a M-dimensional vector.
    
    Returns:
        list: MLE for the mean vector.
    """
    N = len(X)
    mean_mle = [sum(x[i] for x in X) / N for i in range(len(X[0]))]
    return mean_mle


def compute_covariance_mle(X, mean_mle):
    """
    Compute the Maximum Likelihood Estimate for the covariance of a Gaussian distribution.
    
    Parameters:
        X (list of list): List of N data points, each being a M-dimensional vector.
        mean_mle (list): MLE for the mean vector of the data.
    
    Returns:
        list of list: MLE for the covariance matrix (a 2D list).
    """
    N = len(X)
    M = len(X[0])
    covariance_mle = [[0] * M for _ in range(M)]
    
    for x in X:
        diff = [x[i] - mean_mle[i] for i in range(M)]
        for i in range(M):
            for j in range(M):
                covariance_mle[i][j] += diff[i] * diff[j]
    
    covariance_mle = [[covariance_mle[i][j] / N for j in range(M)] for i in range(M)]
    return covariance_mle


# Sequential estimation of mean using Robbins-Monro approach

def sequential_mean_estimation(X):
    """
    Compute the sequential mean estimate using Robbins-Monro algorithm.
    
    Parameters:
        X (list of list): List of N data points, each being a M-dimensional vector.
    
    Returns:
        list: Sequentially updated mean vector.
    """
    N = len(X)
    M = len(X[0])
    mean_sequential = [0] * M
    
    for n, x in enumerate(X, start=1):
        for i in range(M):
            mean_sequential[i] += (x[i] - mean_sequential[i]) / n
    
    return mean_sequential


# Example data points (2D data for simplicity)
X = [[1.2, 2.3], [1.5, 2.7], [1.7, 2.9], [1.3, 2.5]]

# Compute MLE for mean and covariance
mean_mle = compute_mean_mle(X)
covariance_mle = compute_covariance_mle(X, mean_mle)

# Sequential mean estimation
sequential_mean = sequential_mean_estimation(X)

print("Maximum Likelihood Estimate of Mean:", mean_mle)
print("Maximum Likelihood Estimate of Covariance:", covariance_mle)
print("Sequential Mean Estimation:", sequential_mean)
```

# Robbins-Monro Sequential Estimation for MLE

In this notebook, we will explore the Robbins-Monro method for sequential estimation of the Maximum Likelihood Estimation (MLE) of a parameter, particularly focusing on the case of estimating the mean of a Gaussian distribution.

### Robbins-Monro Algorithm

The Robbins-Monro algorithm defines a sequence of estimates of the root $ \theta^* $ as:

$$
\theta^{(N)} = \theta^{(N-1)} + a_N^{-1} z(\theta^{(N-1)})
$$

where $ z(\theta^{(N)}) $ is the observed value of $ z $ when $ \theta $ takes the value $ \theta^{(N)} $. The sequence of coefficients $ \{a_N\} $ are positive numbers that satisfy the following conditions:

1. $ \lim_{N \to \infty} a_N = 0 $
2. $ \sum_{N=1}^{\infty} a_N = \infty $
3. $ \sum_{N=1}^{\infty} a_N^2 < \infty $

These conditions ensure that the algorithm converges to the root with probability one.

### MLE and the Log-Likelihood Function

We are interested in solving the MLE problem, which corresponds to finding the parameter $ \theta_{\text{ML}} $ that maximizes the log-likelihood function. By definition, the MLE solution satisfies:

$$
\frac{1}{N} \sum_{n=1}^{N} \frac{\partial}{\partial \theta} \ln p(x_n | \theta) = 0
$$

Exchanging the derivative and summation, and taking the limit as \( N \to \infty \), we have:

$$
\lim_{N \to \infty} \frac{1}{N} \sum_{n=1}^{N} \frac{\partial}{\partial \theta} \ln p(x_n | \theta) = \mathbb{E}_x \left[ \frac{\partial}{\partial \theta} \ln p(x | \theta) \right]
$$

This shows that the MLE solution corresponds to finding the root of the regression function.

### Applying Robbins-Monro to MLE

For the case of a Gaussian distribution, we can apply the Robbins-Monro method to find the root of the log-likelihood function's derivative. The sequence of estimates for $ \theta^{(N)} $ is given by:

$$
\theta^{(N)} = \theta^{(N-1)} + a_N^{-1} \frac{\partial}{\partial \theta^{(N-1)}} \ln p(x_N | \theta^{(N-1)})
$$

For a Gaussian distribution with mean $ \mu $ and variance $ \sigma^2 $, the derivative of the log-likelihood function with respect to $ \mu $ is:

$$
\frac{\partial}{\partial \mu} \ln p(x | \mu, \sigma^2) = \frac{x - \mu}{\sigma^2}
$$

Thus, the random variable $ z $ becomes:

$$
z = \frac{x - \mu_{\text{ML}}}{\sigma^2}
$$

The expected value of $ z $ defines the regression function, which is a straight line as illustrated in the figure below.

### Example: Sequential Estimation of the Mean of a Gaussian Distribution

For the case of sequential estimation of the mean of a Gaussian distribution, we can use the Robbins-Monro procedure with:

$$
\theta^{(N)} = \mu_{\text{ML}}^{(N)} = \mu_{\text{ML}}^{(N-1)} + \frac{1}{N} \left(x_N - \mu_{\text{ML}}^{(N-1)} \right)
$$

The sequence $ \{ a_N \} $ is chosen such that:

$$
a_N = \frac{\sigma^2}{N}
$$

This ensures that the corrections decrease in magnitude and the process converges to the true mean.

### Convergence Conditions

To ensure convergence, the following conditions on the sequence $ \{a_N\} $ must be satisfied:

1. $ \lim_{N \to \infty} a_N = 0 $
2. $ \sum_{N=1}^{\infty} a_N = \infty $
3. $ \sum_{N=1}^{\infty} a_N^2 < \infty $

These conditions guarantee that the accumulated noise has finite variance and does not spoil convergence to the root.

### Visualizing the Convergence

The figure below shows the regression function for a Gaussian distribution, with the root corresponding to the maximum likelihood estimator $ \mu_{\text{ML}} $.



```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate the Robbins-Monro algorithm for estimating the mean of a Gaussian distribution
def robbins_monro_gaussian(data, initial_guess=0, learning_rate_sequence='inverse'):
    # Initialize the parameters
    theta = initial_guess
    estimates = [theta]
    
    # Iterate through the data points
    for n, x in enumerate(data, 1):
        # Compute the learning rate (aN) based on the chosen sequence
        if learning_rate_sequence == 'inverse':
            aN = 1 / n  # Inverse sequence for learning rate
        elif learning_rate_sequence == 'constant':
            aN = 0.1  # Constant learning rate
        else:
            raise ValueError("Invalid learning rate sequence. Choose 'inverse' or 'constant'.")
        
        # Update the estimate using the Robbins-Monro update rule
        # The derivative of the log likelihood function for the Gaussian is (x - mu) / sigma^2
        # Here, sigma is assumed to be 1 for simplicity
        derivative = (x - theta) / 1  # Simplifying to sigma^2 = 1
        theta = theta + aN * derivative
        
        # Store the estimate
        estimates.append(theta)
    
    return np.array(estimates)

# Simulating data from a Gaussian distribution with mean 5 and standard deviation 2
np.random.seed(42)  # For reproducibility
true_mean = 5
sigma = 2
n_samples = 100
data = np.random.normal(true_mean, sigma, n_samples)

# Run the Robbins-Monro algorithm on the simulated data
initial_guess = 0  # Starting guess for the mean
estimates = robbins_monro_gaussian(data, initial_guess, learning_rate_sequence='inverse')

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(estimates, label='Estimated Mean', color='blue')
plt.axhline(y=true_mean, color='red', linestyle='--', label=f'True Mean = {true_mean}')
plt.title('Robbins-Monro Algorithm for Estimating the Mean of a Gaussian Distribution')
plt.xlabel('Iteration')
plt.ylabel('Estimated Mean')
plt.legend()
plt.grid(True)
plt.show()
```

## Bayesian Inference for the Mean of a Gaussian Distribution

In this section, we will develop a Bayesian treatment for inferring the mean $ \mu $ of a Gaussian distribution, assuming the variance $ \sigma^2 $ is known. Given a set of $ N $ observations $ X = \{ x_1, x_2, \dots, x_N \} $, we aim to infer $ \mu $.

### Likelihood Function

The likelihood function, which represents the probability of observing the data given $ \mu $, is given by:

$$
p(X|\mu) = \prod_{n=1}^{N} p(x_n|\mu) = \prod_{n=1}^{N} \frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( - \frac{(x_n - \mu)^2}{2 \sigma^2} \right)
$$

This is the product of the Gaussian probability density functions for each data point, where $ \mu $ is the parameter we wish to infer.

### Prior Distribution

We assume a prior distribution for $ \mu $, which is also Gaussian:

$$
p(\mu) = \mathcal{N}(\mu | \mu_0, \sigma_0^2)
$$

where $ \mu_0 $ is the prior mean and $ \sigma_0^2 $ is the prior variance.

### Posterior Distribution

Using Bayes' theorem, the posterior distribution is proportional to the product of the likelihood and the prior:

$$
p(\mu|X) \propto p(X|\mu) p(\mu)
$$

After completing the square in the exponent, we find that the posterior distribution is also Gaussian:

$$
p(\mu|X) = \mathcal{N}(\mu | \mu_N, \sigma_N^2)
$$

where the posterior mean \( \mu_N \) and posterior variance \( \sigma_N^2 \) are given by:

$$
\mu_N = \frac{\sigma_0^2 \mu_0 + N \sigma^2 \bar{x}}{\sigma_0^2 + N \sigma^2}
$$

$$
\sigma_N^2 = \frac{\sigma^2}{\frac{1}{\sigma_0^2} + \frac{N}{\sigma^2}}
$$

where $ \bar{x} $ is the sample mean of the data:

$$
\bar{x} = \frac{1}{N} \sum_{n=1}^{N} x_n
$$

### Interpretation of Posterior

- **Posterior Mean**: The posterior mean $ \mu_N $ is a compromise between the prior mean $ \mu_0 $ and the maximum likelihood estimate $ \mu_{\text{ML}} $. If no data is observed ($ N = 0 $), the posterior mean is equal to the prior mean. As the number of data points increases, the posterior mean converges to the maximum likelihood estimate.

- **Posterior Variance**: The posterior variance $ \sigma_N^2 $ decreases as the number of observations increases. With a large number of data points, the variance becomes very small, and the posterior distribution becomes sharply peaked around the maximum likelihood estimate.

### Maximum Likelihood Estimate

The maximum likelihood estimate $ \mu_{\text{ML}} $ is simply the sample mean:

$$
\mu_{\text{ML}} = \frac{1}{N} \sum_{n=1}^{N} x_n
$$

In the Bayesian framework, as $ N \to \infty $, the posterior mean approaches the maximum likelihood estimate.

### Sequential Inference

The Bayesian framework allows for a sequential update of the posterior as new data points are observed. Given the posterior distribution after $ N-1 $ data points, the posterior after observing the $ N $-th data point is:

$$
p(\mu|X) \propto p(\mu|X_{N-1}) p(x_N|\mu)
$$

This sequential view is particularly useful for problems in which data is collected over time, and we wish to update our belief about $ \mu $ as new data becomes available.

---

## Bayesian Inference for the Variance of a Gaussian Distribution

Now, let's consider the problem of inferring the variance $ \sigma^2 $ of a Gaussian distribution, assuming that the mean $ \mu $ is known. To make the calculations simpler, we work with the precision $ \lambda = \frac{1}{\sigma^2} $.

### Likelihood Function for Precision

The likelihood function for $ \lambda $ is given by:

$$
p(X|\lambda) \propto \lambda^{N/2} \exp \left( -\frac{\lambda}{2} \sum_{n=1}^{N} (x_n - \mu)^2 \right)
$$

### Prior Distribution for Precision

We assume a prior for $ \lambda $ that follows a Gamma distribution:

$$
p(\lambda) = \text{Gam}(\lambda | a_0, b_0) \propto \lambda^{a_0 - 1} \exp(-b_0 \lambda)
$$

where $ a_0 $ and $ b_0 $ are the parameters of the Gamma distribution.

### Posterior Distribution

Multiplying the likelihood and the prior, we obtain the posterior distribution:

$$
p(\lambda | X) \propto \lambda^{a_0 + N/2 - 1} \exp \left( - \left( b_0 + \frac{1}{2} \sum_{n=1}^{N} (x_n - \mu)^2 \right) \lambda \right)
$$

This is another Gamma distribution with updated parameters:

$$
\lambda | X \sim \text{Gam} \left( a_N, b_N \right)
$$

where the updated parameters are:

$$
a_N = a_0 + \frac{N}{2}
$$

$$
b_N = b_0 + \frac{1}{2} \sum_{n=1}^{N} (x_n - \mu)^2
$$

### Interpretation

- The posterior distribution for $ \lambda $ is a Gamma distribution, which is conjugate to the likelihood.
- As more data points are observed, the precision (inverse variance) of the posterior distribution increases, resulting in a more certain estimate of the variance.

---

This completes the Bayesian analysis for both the mean and variance of a Gaussian distribution.

## Bayesian Inference for the Gaussian

From equation (2.150), we observe that the effect of observing \(N\) data points is to increase the value of the coefficient $a$ by $N/2$. Thus, we can interpret the parameter $a_0$ in the prior as representing $2a_0$ "effective" prior observations. Similarly, from equation (2.151), we see that the $N$ data points contribute $N \cdot \sigma_{\text{ML}}^2 / 2$ to the parameter $b$, where $\sigma_{\text{ML}}^2$ is the variance. This means we can interpret the parameter $b_0$ in the prior as arising from the $2a_0$ "effective" prior observations, which have variance $2b_0 / 2a_0 = b_0 / a_0$. This interpretation is similar to the Dirichlet prior.

These distributions are examples of the exponential family, and we will see that the interpretation of a conjugate prior in terms of effective fictitious data points is a general concept for the exponential family of distributions. Instead of working with precision, we can consider the variance itself. The conjugate prior in this case is called the inverse gamma distribution, though we will not discuss it further because it is more convenient to work with the precision.

### Case 1: Both the Mean and Precision are Unknown

Now, suppose both the mean $\mu$ and the precision $\lambda$ are unknown. To find a conjugate prior, we consider the likelihood function's dependence on $\mu$ and $\lambda$:

$$
p(X|\mu, \lambda) \propto \lambda^{1/2} \exp \left( - \frac{1}{2} \lambda \sum_{n=1}^{N} (x_n - \mu)^2 \right)
$$

We seek a prior distribution $p(\mu, \lambda)$ that has the same functional dependence on $\mu$ and $\lambda$ as the likelihood. This suggests a prior of the form:

$$
p(\mu, \lambda) \propto \lambda^{1/2} \exp \left( - \frac{1}{2} \left( c \lambda \mu^2 - d \lambda \right) \right)
$$

By inspection, we find that:

$$
p(\mu | \lambda) = N(\mu | \mu_0, (\beta \lambda)^{-1}), \quad p(\lambda) = \text{Gamma}(\lambda | a, b)
$$

Thus, the normalized prior takes the form:

$$
p(\mu, \lambda) = N(\mu | \mu_0, (\beta \lambda)^{-1}) \, \text{Gamma}(\lambda | a, b)
$$

where the constants are defined as:

$$
\mu_0 = \frac{c}{\beta}, \quad a = 1 + \frac{\beta}{2}, \quad b = d - \frac{c^2}{2 \beta}
$$

This distribution is called the **normal-gamma** or **Gaussian-gamma** distribution. It is not simply the product of independent Gaussian and Gamma distributions, as the precision of $\mu$ is a linear function of $\lambda$. Even if we choose independent priors for $\mu$ and $\lambda$, the posterior will exhibit a coupling between the precision of $\mu$ and the value of $\lambda$.

### Multivariate Case

For the multivariate Gaussian distribution $N(x | \mu, \Lambda^{-1})$ with $D$-dimensional variable $x$, the conjugate prior for the mean $\mu$, assuming the precision $\Lambda$ is known, is again a Gaussian. For a known mean and an unknown precision matrix $\Lambda$, the conjugate prior is the **Wishart distribution**:

$$
W(\Lambda | W, \nu) = B | \Lambda |^{(\nu - D - 1)/2} \exp \left( - \frac{1}{2} \text{Tr}(W \Lambda^{-1}) \right)
$$

where $\nu$ is the degrees of freedom, $W$ is a $D \times D$ scale matrix, and $\text{Tr}(\cdot)$ denotes the trace. The normalization constant $B$ is given by:

$$
B(W, \nu) = |W|^{-\nu/2} \frac{\Gamma_D(\nu/2)}{\Gamma_D(\nu/2 + D/2)} \pi^{D(D-1)/4} \, .
$$

The joint conjugate prior for both the mean $\mu$ and precision $\Lambda$ in the multivariate case is given by the **normal-Wishart** distribution:

$$
p(\mu, \Lambda | \mu_0, \beta, W, \nu) = N(\mu | \mu_0, (\beta \Lambda)^{-1}) W(\Lambda | W, \nu)
$$

### Student's t-Distribution

We have seen that the conjugate prior for the precision of a Gaussian is the Gamma distribution. When we have a univariate Gaussian $N(x | \mu, \tau^{-1})$ and a Gamma prior $\text{Gam}(\tau | a, b)$, and we integrate out the precision, we obtain the marginal distribution of $x$:

$$
p(x | \mu, a, b) = \int_0^\infty N(x | \mu, \tau^{-1}) \, \text{Gam}(\tau | a, b) \, d\tau
$$

This results in the **Student's t-distribution**:

$$
\text{St}(x | \mu, \lambda, \nu) = \frac{\Gamma(\frac{\nu + 1}{2})}{\Gamma(\frac{\nu}{2}) \sqrt{\pi \nu \lambda}} \left( 1 + \frac{(x - \mu)^2}{\nu \lambda} \right)^{-\frac{\nu + 1}{2}}
$$

Here, $\lambda$ is the precision of the t-distribution and $\nu$ is the degrees of freedom. The effect of the parameter $\nu$ is illustrated in the plot below:

### Robustness of Students t-distribution Compared to Gaussian Distribution

Fig.16 illustrates the robustness of the Students t-distribution when handling outliers compared to the Gaussian distribution. In many practical scenarios, outliers may arise either from heavy-tailed distributions or due to mislabeled data. This robustness is particularly useful in regression, as least squares regression lacks robustness due to its reliance on maximum likelihood estimation under a Gaussian assumption.

### Mathematical Representation of the Students t-distribution

To derive the Students t-distribution, we start with the marginalization over the precision parameter $\tau$ in a Gaussian distribution with a Gamma prior. Given a Gaussian distribution $ N(x|\mu, \tau^{-1}) $ with a Gamma prior $ \text{Gam}(\tau | a, b) $, we can integrate out $\tau$ to obtain the marginal distribution for $ x $:

$$
p(x|\mu, a, b) = \int_0^{\infty} N(x|\mu, \tau^{-1}) \, \text{Gam}(\tau|a, b) \, d\tau
$$

By reparameterizing, let $ \nu = 2a $, $ \lambda = \frac{a}{b} $, and $ \eta = \frac{\tau b}{a} $. The Students t-distribution can then be expressed as:

$$
St(x|\mu, \lambda, \nu) = \int_0^{\infty} N(x|\mu, (\eta \lambda)^{-1}) \, \text{Gam}(\eta|\nu/2, \nu/2) \, d\eta
$$

### Multivariate Students t-distribution

For a multivariate Gaussian \( N(x|\mu, \Lambda) \), we generalize to the multivariate Students t-distribution as follows:

$$
St(x|\mu, \Lambda, \nu) = \int_0^{\infty} N(x|\mu, (\eta \Lambda)^{-1}) \, \text{Gam}(\eta|\nu/2, \nu/2) \, d\eta
$$

Evaluating this integral, we obtain:

$$
St(x|\mu, \Lambda, \nu) = \frac{\Gamma \left(\frac{D}{2} + \frac{\nu}{2}\right)}{\Gamma \left(\frac{\nu}{2}\right) (\pi \nu)^{D/2} |\Lambda|^{1/2}} \left(1 + \frac{\Delta^2}{\nu}\right)^{-\frac{D}{2} - \frac{\nu}{2}}
$$

where \( D \) is the dimensionality of \( x \), and \( \Delta^2 \) is the squared Mahalanobis distance, defined as:

$$
\Delta^2 = (x - \mu)^T \Lambda (x - \mu)
$$

### Properties of the Multivariate Students t-distribution

The multivariate Students t-distribution has the following properties:

- **Expectation**:
  $$
  \mathbb{E}[x] = \mu, \quad \text{if } \nu > 1
  $$

- **Covariance**:
  $$
  \text{cov}[x] = \frac{\Lambda^{-1}}{\nu - 2}, \quad \text{if } \nu > 2
  $$

- **Mode**:
  $$
  \text{mode}[x] = \mu
  $$

These properties highlight the robustness of the Students t-distribution, as it provides a distribution with heavier tails than the Gaussian, making it less sensitive to outliers.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Set random seed for reproducibility
np.random.seed(0)

# Generate sample data from a Gaussian distribution
data = np.random.normal(loc=0, scale=1, size=30)

# Add outliers to demonstrate robustness of the t-distribution
data_with_outliers = np.append(data, [8, 9, 10])

# Fit Gaussian and Student's t-distributions
mu_gaussian, std_gaussian = norm.fit(data_with_outliers)
df, mu_t, std_t = t.fit(data_with_outliers)

# Generate a range of values for plotting
x = np.linspace(-5, 15, 500)

# Compute PDFs for both distributions
pdf_gaussian = norm.pdf(x, mu_gaussian, std_gaussian)
pdf_t = t.pdf(x, df, mu_t, std_t)

# Plot histogram and fitted distributions
plt.figure(figsize=(10, 6))
plt.hist(data_with_outliers, bins=15, density=True, alpha=0.5, color='lightgrey', edgecolor='black', label="Data with Outliers")
plt.plot(x, pdf_gaussian, 'g-', label="Gaussian Fit", linewidth=2)
plt.plot(x, pdf_t, 'r-', label="Student's t Fit", linewidth=2)
plt.title("Robustness of Students t-distribution Compared to Gaussian")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
