### Representation of the Mean for Periodic Variables



Fig.17 Illustration of the representation of val- ues n of a periodic variable as two- dimensional vectors xn living on the unit circle. Also shown is the average x of those vectors.

The mean of a set of observations $ D = \{\theta_1, \dots, \theta_N\} $ of a periodic variable is calculated as:

$$
\mathbf{x} = \frac{1}{N} \sum_{n=1}^N \mathbf{x}_n
$$

where each observation $ \theta_n $ is represented as a point on the unit circle with Cartesian coordinates $ \mathbf{x}_n = (\cos \theta_n, \sin \theta_n) $.

The average vector $ \mathbf{x} $ has Cartesian coordinates given by:

$$
\mathbf{x} = (r \cos \theta, r \sin \theta)
$$

where:

$$
r \cos \theta = \frac{1}{N} \sum_{n=1}^N \cos \theta_n, \quad r \sin \theta = \frac{1}{N} \sum_{n=1}^N \sin \theta_n
$$

The angle $ \theta $ of the average vector $ \mathbf{x} $ is then:

$$
\theta = \tan^{-1} \left( \frac{\sum_{n=1}^N \sin \theta_n}{\sum_{n=1}^N \cos \theta_n} \right)
$$

---

### Conditions for a Periodic Probability Density

To model a periodic variable $ \theta $ with a probability density function $ p(\theta) $, the density must satisfy these conditions:

1. **Non-negativity**:
   $$
   p(\theta) \geq 0
   $$

2. **Normalization**:
   $$
   \int_0^{2\pi} p(\theta) \, d\theta = 1
   $$

3. **Periodicity**:
   $$
   p(\theta + 2\pi) = p(\theta)
   $$

From condition (3), it follows that $ p(\theta + M \cdot 2\pi) = p(\theta) $ for any integer $ M $.

---

### Deriving a Gaussian-like Periodic Distribution

To obtain a Gaussian-like distribution that satisfies these properties, consider a two-dimensional Gaussian distribution over variables $ x = (x_1, x_2) $ with mean $ \mu = (\mu_1, \mu_2) $ and covariance matrix $ \Sigma = \sigma^2 I $ (where $ I $ is the $ 2 \times 2 $ identity matrix):

$$
p(x_1, x_2) = \frac{1}{2 \pi \sigma^2} \exp \left( -\frac{(x_1 - \mu_1)^2 + (x_2 - \mu_2)^2}{2 \sigma^2} \right)
$$

This distribution has circular contours of constant \( p(x) \), as illustrated in Figure 2.18.

Now, consider the value of this distribution along a circle of fixed radius. By construction, this distribution will be periodic, though it wont be normalized.

---

### Transformation to Polar Coordinates

Transform the Cartesian coordinates $ (x_1, x_2) $ to polar coordinates $ (r, \theta) $:

$$
x_1 = r \cos \theta, \quad x_2 = r \sin \theta
$$

Similarly, express the mean \( \mu \) in polar coordinates as:

$$
\mu_1 = r_0 \cos \theta_0, \quad \mu_2 = r_0 \sin \theta_0
$$

---

### Focusing on the Exponent

Substitute these transformations into the two-dimensional Gaussian and condition on the unit circle $ r = 1 $. The exponent of the Gaussian then becomes:

$$
-\frac{1}{2 \sigma^2} \left( (r \cos \theta - r_0 \cos \theta_0)^2 + (r \sin \theta - r_0 \sin \theta_0)^2 \right)
$$

Expanding and simplifying using trigonometric identities, we get:

$$
-\frac{1}{2 \sigma^2} \left( 1 + r_0^2 - 2 r_0 \cos(\theta - \theta_0) \right) = \frac{r_0}{\sigma^2} \cos(\theta - \theta_0) + \text{const}
$$

Defining $ m = \frac{r_0}{\sigma^2} $, we obtain the form:

$$
p(\theta | \theta_0, m) = \frac{1}{2\pi I_0(m)} \exp \{ m \cos(\theta - \theta_0) \}
$$

where $ I_0(m) $ is the zeroth-order modified Bessel function of the first kind:

$$
I_0(m) = \frac{1}{2 \pi} \int_0^{2 \pi} \exp \{ m \cos \theta \} \, d\theta
$$

This defines the **von Mises distribution**, which is the Gaussian-like distribution on a unit circle, with $ \theta_0 $ as the mean direction and $ m $ as the concentration parameter (analogous to the inverse variance in the Gaussian distribution).


### Von Mises Distribution



Fig.18 The von Mises distribution can be derived by considering a two-dimensional Gaussian of the form (2.173), whose density contours are shown in blue and conditioning on the unit circle shown in red.

The von Mises distribution is defined by:

$$
p(\theta | \theta_0, m) = \frac{1}{2 \pi I_0(m)} \exp \{ m \cos(\theta - \theta_0) \}
$$

where:
- $ \theta_0 $ is the mean angle of the distribution,
- $ m $ is the concentration parameter (analogous to the inverse variance in a Gaussian),
- $ I_0(m) $ is the zeroth-order modified Bessel function of the first kind:

$$
I_0(m) = \frac{1}{2 \pi} \int_0^{2 \pi} \exp \{ m \cos \theta \} \, d\theta
$$

For large $ m $, the von Mises distribution approximates a Gaussian distribution.

---

### Log Likelihood for Maximum Likelihood Estimation of the von Mises Distribution


Fig.19 The von Mises distribution plotted for two different parameter values, shown as a Cartesian plot on the left and as the corresponding polar plot on the right.

The log likelihood function for $ N $ observations $ D = \{\theta_1, \dots, \theta_N\} $ is given by:

$$
\ln p(D | \theta_0, m) = -N \ln (2 \pi) - N \ln I_0(m) + m \sum_{n=1}^N \cos(\theta_n - \theta_0)
$$

### The von Mises Distribution

The von Mises distribution, plotted in Figure 2.19, is shown for two different parameter values as a Cartesian plot on the left and a polar plot on the right.

Using the trigonometric identities:

$$
\cos^2(A) + \sin^2(A) = 1
$$

and

$$
\cos(A) \cos(B) + \sin(A) \sin(B) = \cos(A - B),
$$

we can simplify the expression for the von Mises distribution.

If we define $ m = \frac{r_0}{\sigma^2} $, we obtain the final expression for the distribution $ p(\theta) $ along the unit circle $ r = 1 $:

$$
p(\theta | \theta_0, m) = \frac{1}{2\pi I_0(m)} \exp \{ m \cos(\theta - \theta_0) \}
$$

where the von Mises distribution (or circular normal distribution) is represented. Here, $ \theta_0 $ is the mean of the distribution, and $ m $ (known as the concentration parameter) is analogous to the inverse variance (precision) of a Gaussian. The normalization coefficient involves $ I_0(m) $, the zeroth-order Bessel function of the first kind (Abramowitz and Stegun, 1965), defined by:

$$
I_0(m) = \frac{1}{2\pi} \int_{0}^{2\pi} \exp \{ m \cos \theta \} \, d\theta.
$$

For large $ m $, the distribution approximates a Gaussian.

---

### Maximum Likelihood Estimators for the von Mises Distribution

The maximum likelihood estimators for $ \theta_0 $ and $ m $ in the von Mises distribution can be derived from the log-likelihood function:

$$
\ln p(D | \theta_0, m) = -N \ln(2\pi) - N \ln I_0(m) + m \sum_{n=1}^{N} \cos(\theta_n - \theta_0).
$$


### Plot of the Bessel Function $ I_0(m)$  and the Function $ A(m) $

The Bessel function $ I_0(m) $, defined by equation (2.180), and the function $ A(m) $, defined by equation (2.186), are plotted in Fig.20.

To set the derivative with respect to $ \theta_0 $ to zero, we get:

$$
\sum_{n=1}^{N} \sin(\theta_n - \theta_0) = 0.
$$

To solve for $ \theta_0 $, we use the trigonometric identity:

$$
\sin(A - B) = \cos(B) \sin(A) - \sin(B) \cos(A)
$$

which leads to the maximum likelihood solution:

$$
\theta_0^{\text{ML}} = \tan^{-1} \left( \frac{\sum_n \sin \theta_n}{\sum_n \cos \theta_n} \right)
$$

This expression is the same as equation (2.169), which represents the mean of observations viewed in a two-dimensional Cartesian space.

### Maximizing with Respect to $ m $

Maximizing the log likelihood function with respect to $ m $, and using the identity $ I_0'(m) = I_1(m) $ (from Abramowitz and Stegun, 1965), we obtain:

$$
A(m) = \frac{1}{N} \sum_{n=1}^{N} \cos(\theta_n - \theta_0^{\text{ML}})
$$

where $ \theta_0^{\text{ML}} $ is the maximum likelihood estimate of $ \theta_0 $, and we define:

$$
A(m) = \frac{I_1(m)}{I_0(m)}

A(m^{\text{ML}}) = \frac{1}{N} \sum_{n=1}^{N} \left( \cos \theta_n \cos \theta_0^{\text{ML}} + \sin \theta_n \sin \theta_0^{\text{ML}} \right)

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the zeroth-order Bessel function of the first kind (I0)
def bessel_I0(m, n_terms=50):
    """
    Compute the zeroth-order Bessel function of the first kind, I0(m),
    using a series expansion approximation.
    """
    I0 = 0
    for k in range(n_terms):
        I0 += (1 / math.factorial(k))**2 * (m / 2)**(2 * k)
    return I0

# Step 2: Define the von Mises distribution
def von_mises_pdf(theta, theta_0, m):
    """
    Probability density function for the von Mises distribution.
    
    Parameters:
    - theta : angle in radians
    - theta_0 : mean angle in radians
    - m : concentration parameter (analogous to inverse variance)
    """
    return (1 / (2 * math.pi * bessel_I0(m))) * math.exp(m * math.cos(theta - theta_0))

# Step 3: Plot the von Mises distribution for different parameters
theta_values = np.linspace(0, 2 * math.pi, 100)
m_values = [1, 5]
theta_0_values = [math.pi / 4, 3 * math.pi / 4]

plt.figure(figsize=(12, 6))

for i, m in enumerate(m_values):
    for j, theta_0 in enumerate(theta_0_values):
        # Compute von Mises distribution values
        pdf_values = [von_mises_pdf(theta, theta_0, m) for theta in theta_values]
        
        # Plot Cartesian plot
        plt.subplot(2, 2, i * 2 + j + 1)
        plt.plot(theta_values, pdf_values, label=f'm={m}, ={theta_0:.2f}')
        plt.title(f'Von Mises Distribution (m={m}, ={theta_0:.2f})')
        plt.xlabel(' (radians)')
        plt.ylabel('p()')
        plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Maximum Likelihood Estimation (MLE) for theta_0 and m
def compute_mle_theta_0(data_angles):
    """
    Compute MLE for  using the given sample angles.
    
    Parameters:
    - data_angles : list of angles () in radians
    
    Returns:
    - theta_0_mle : MLE for 
    """
    sin_sum = sum(math.sin(theta) for theta in data_angles)
    cos_sum = sum(math.cos(theta) for theta in data_angles)
    theta_0_mle = math.atan2(sin_sum, cos_sum)
    return theta_0_mle

def compute_mle_m(data_angles, theta_0_mle, n_terms=50):
    """
    Compute MLE for m using the given sample angles.
    
    Parameters:
    - data_angles : list of angles () in radians
    - theta_0_mle : MLE for 
    - n_terms : number of terms for Bessel function approximation
    
    Returns:
    - m_mle : MLE for m
    """
    N = len(data_angles)
    cos_term = sum(math.cos(theta - theta_0_mle) for theta in data_angles) / N
    # A(m) = I1(m) / I0(m), and we need to solve A(m)  cos_term
    # Here, we approximate m by finding the best m such that A(m)  cos_term.
    
    # Numerical search for m
    m_mle = 0
    for m_candidate in np.linspace(0, 10, 1000):  # Use a large range to find the best m
        A_m_candidate = bessel_I0(m_candidate + 1e-5, n_terms) / bessel_I0(m_candidate, n_terms)
        if abs(A_m_candidate - cos_term) < 1e-3:  # tolerance level for matching A(m)  cos_term
            m_mle = m_candidate
            break
    return m_mle

# Example Data for MLE Calculation
data_angles = np.random.vonmises(math.pi / 4, 2, size=100)

# Compute MLE for theta_0 and m
theta_0_mle = compute_mle_theta_0(data_angles)
m_mle = compute_mle_m(data_angles, theta_0_mle)

print("MLE for :", theta_0_mle)
print("MLE for m:", m_mle)
```

### Mixture of Gaussians

The Gaussian distribution has some analytical properties, but it may not always be ideal for modeling real-world data. For instance, consider the 'Old Faithful' dataset in Fi.21, which includes 272 measurements of the eruption of the Old Faithful geyser at Yellowstone National Park, USA. Each measurement consists of the eruption duration and waiting time.

#### Mixture of Three Gaussians in Two-Dimensional Space
In Fig.23, we illustrate a mixture of three Gaussians in a two-dimensional space:
- **(a)** Contours of constant density for each mixture component, with red, blue, and green representing the three components. The mixing coefficients for each component are shown below.
- **(b)** Contours of the marginal probability density $ p(x) $ of the mixture distribution.
- **(c)** A surface plot of the distribution $ p(x) $.

We can express the marginal density as:

$$
p(x) = \sum_{k=1}^K p(k) p(x|k)
$$

This is equivalent to

$$
\pi_k = p(k)
$$

where $ p(k) $ is the prior probability of selecting the $ k $-th component, and $ N(x | \mu_k, \Sigma_k) = p(x|k) $ is the probability of $ x $ conditioned on $ k $. An important aspect is the posterior probabilities $ p(k | x) $, also known as *responsibilities*, given by Bayes' theorem as follows:

$$
\gamma_k(x) \equiv p(k|x) = \frac{p(k) p(x|k)}{\sum_{l} p(l) p(x|l)}
$$

or equivalently,

$$
\gamma_k(x) = \frac{\pi_k N(x | \mu_k, \Sigma_k)}{\sum_{l} \pi_l N(x | \mu_l, \Sigma_l)}
$$

The parameters of the Gaussian mixture distribution are governed by $ \pi $, $ \mu $, and $ \Sigma $, where we use the notation $ \pi = \{\pi_1, \dots, \pi_K\} $, $ \mu = \{\mu_1, \dots, \mu_K\} $, and $ \Sigma = \{\Sigma_1, \dots, \Sigma_K\} $.

One method to determine these parameter values is to use maximum likelihood estimation. The log-likelihood function is given by:

$$
\ln p(X | \pi, \mu, \Sigma) = \sum_{n=1}^N \ln \left( \sum_{k=1}^K \pi_k N(x_n | \mu_k, \Sigma_k) \right)
$$

where $ X = \{x_1, \dots, x_N\} $. Due to the summation over $ k $ inside the logarithm, the maximum likelihood solution does not have a closed-form analytical solution. Thus, iterative numerical optimization techniques are commonly used (e.g., Fletcher, 1987; Nocedal and Wright, 1999; Bishop and Nabney, 2008). Alternatively, the expectation-maximization (EM) framework, discussed in Chapter 9, is often employed for this purpose.




Fig.23 Illustration of a mixture of 3 Gaussians in a two-dimensional space. (a) Contours of constant density for each of the mixture components, in which the 3 components are denoted red, blue and green, and the values of the mixing coefcients are shown below each component. (b) Contours of the marginal probability density p(x) of the mixture distribution. (c) A surface plot of the distribution p(x).

```python
import random
import math

# Initialize parameters
class GaussianMixtureModel:
    def __init__(self, k=2, iterations=100):
        self.k = k  # Number of Gaussian components
        self.iterations = iterations
        self.means = [random.random() for _ in range(k)]
        self.variances = [1.0 for _ in range(k)]
        self.weights = [1.0 / k for _ in range(k)]
    
    # Gaussian probability density function
    def gaussian(self, x, mean, variance):
        exponent = math.exp(-((x - mean) ** 2) / (2 * variance))
        return (1 / math.sqrt(2 * math.pi * variance)) * exponent
    
    # E-step: calculate responsibilities
    def expectation(self, data):
        responsibilities = []
        for x in data:
            total = 0
            r = []
            for j in range(self.k):
                p = self.weights[j] * self.gaussian(x, self.means[j], self.variances[j])
                r.append(p)
                total += p
            responsibilities.append([p / total for p in r])
        return responsibilities
    
    # M-step: update parameters based on responsibilities
    def maximization(self, data, responsibilities):
        for j in range(self.k):
            weight_sum = sum([r[j] for r in responsibilities])
            self.weights[j] = weight_sum / len(data)
            
            self.means[j] = sum([r[j] * x for r, x in zip(responsibilities, data)]) / weight_sum
            variance_sum = sum([r[j] * ((x - self.means[j]) ** 2) for r, x in zip(responsibilities, data)])
            self.variances[j] = variance_sum / weight_sum
    
    # Fit model to data
    def fit(self, data):
        for _ in range(self.iterations):
            responsibilities = self.expectation(data)
            self.maximization(data, responsibilities)
    
    # Predict probability density for new data point
    def predict(self, x):
        return sum([self.weights[j] * self.gaussian(x, self.means[j], self.variances[j]) for j in range(self.k)])

# Generate synthetic data for testing
def generate_data():
    data = []
    for _ in range(100):
        if random.random() > 0.5:
            data.append(random.gauss(5, 1))  # Gaussian centered at 5
        else:
            data.append(random.gauss(10, 1)) # Gaussian centered at 10
    return data

# Run the GMM on synthetic data
data = generate_data()
gmm = GaussianMixtureModel(k=2, iterations=20)
gmm.fit(data)

# Print results
print("Means:", gmm.means)
print("Variances:", gmm.variances)
print("Weights:", gmm.weights)
```
