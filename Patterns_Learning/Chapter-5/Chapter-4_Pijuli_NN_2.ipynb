##  The Hessian Matrix

We have shown how the technique of backpropagation can be used to obtain the first derivatives of an error function with respect to the weights in the network. Backpropagation can also be used to evaluate the second derivatives of the error, given by:

$$
\frac{\partial^2 E}{\partial w_{ji} \partial w_{lk}} \tag{5.78}
$$

Note that it is sometimes convenient to consider all of the weight and bias parameters as elements $ w_i $ of a single vector, denoted $ w $, in which case the second derivatives form the elements $ H_{ij} $ of the Hessian matrix $ H $, where $ i, j \in \{1, \dots, W\} $ and $ W $ is the total number of weights and biases. The Hessian plays an important role in many aspects of neural computing, including the following:

1. Several nonlinear optimization algorithms used for training neural networks are based on considerations of the second-order properties of the error surface, which are controlled by the Hessian matrix (Bishop and Nabney, 2008).
2. The Hessian forms the basis of a fast procedure for re-training a feed-forward network following a small change in the training data (Bishop, 1991).
3. The inverse of the Hessian has been used to identify the least significant weights in a network as part of network ‘pruning’ algorithms (Le Cun et al., 1990).
4. The Hessian plays a central role in the Laplace approximation for a Bayesian neural network (see Section 5.7). Its inverse is used to determine the predictive distribution for a trained network, its eigenvalues determine the values of hyperparameters, and its determinant is used to evaluate the model evidence.

Various approximation schemes have been used to evaluate the Hessian matrix for a neural network. However, the Hessian can also be calculated exactly using an extension of the backpropagation technique. An important consideration for many applications of the Hessian is the efficiency with which it can be evaluated. If there are $ W $ parameters (weights and biases) in the network, then the Hessian matrix has dimensions $ W \times W $, and so the computational effort needed to evaluate the Hessian will scale like $ O(W^2) $ for each pattern in the data set. As we shall see, there are efficient methods for evaluating the Hessian whose scaling is indeed $ O(W^2) $.

###  Diagonal Approximation

Some of the applications for the Hessian matrix discussed above require the inverse of the Hessian, rather than the Hessian itself. For this reason, there has been some interest in using a diagonal approximation to the Hessian, in other words, one that simply replaces the off-diagonal elements with zeros, because its inverse is trivial to evaluate. Again, we shall consider an error function $ E $ that consists of a sum of terms, one for each pattern in the data set, so that:

$$
E = \sum_{n} E_n
$$

The Hessian can then be obtained by considering one pattern at a time, and then summing the results over all patterns. From (5.48), the diagonal elements of the Hessian for pattern $ n $ can be written:

$$
\frac{\partial^2 E_n}{\partial w_{ji}^2} = \frac{\partial^2 E_n}{\partial a_j^2} \tag{5.79}
$$

Using (5.48) and (5.49), the second derivatives on the right-hand side of (5.79) can be found recursively using the chain rule of differential calculus to give a backpropagation equation of the form:

$$
\frac{\partial^2 E_n}{\partial a_j^2} = h \left( \frac{\partial^2 E_n}{\partial a_j} \right)^2 w_{kj} w_k + h \left( \frac{\partial E_n}{\partial a_j} \right) w_{kj} \tag{5.80}
$$

If we now neglect off-diagonal elements in the second-derivative terms, we obtain (Becker and Le Cun, 1989; Le Cun et al., 1990):

$$
\frac{\partial^2 E_n}{\partial a_j^2} = h \left( \frac{\partial^2 E_n}{\partial a_j} \right)^2 w_{kj}^2 + h \left( \frac{\partial E_n}{\partial a_j} \right) w_{kj} \tag{5.81}
$$

Note that the number of computational steps required to evaluate this approximation is $ O(W) $, where $ W $ is the total number of weight and bias parameters in the network, compared with $ O(W^2) $ for the full Hessian. 

The major problem with diagonal approximations, however, is that in practice the Hessian is typically found to be strongly nondiagonal, and so these approximations, which are driven mainly by computational convenience, must be treated with care.

###  Outer Product Approximation

When neural networks are applied to regression problems, it is common to use a sum-of-squares error function of the form:

$$
E = \frac{1}{2} \sum_{n=1}^{N} (y_n - t_n)^2 \tag{5.82}
$$

where we have considered the case of a single output in order to keep the notation simple (the extension to several outputs is straightforward). We can then write the Hessian matrix in the form:

$$
H = \nabla \nabla E = \sum_{n=1}^{N} \nabla y_n \nabla y_n + (y_n - t_n) \nabla \nabla y_n \tag{5.83}
$$

If the network has been trained on the data set, and its outputs $ y_n $ happen to be very close to the target values $ t_n $, then the second term in (5.83) will be small and can be neglected. More generally, however, it may be appropriate to neglect this term by the following argument. Recall from Section 1.5.5 that the optimal function that minimizes a sum-of-squares loss is the conditional average of the target data. The quantity $ (y_n - t_n) $ is then a random variable with zero mean. If we assume that its value is uncorrelated with the value of the second derivative term on the right-hand side of (5.83), then the whole term will average to zero in the summation over $ n $. By neglecting the second term in (5.83), we arrive at the Levenberg–Marquardt approximation or outer product approximation (because the Hessian matrix is built up from a sum of outer products of vectors), given by:

$$
H \approx \sum_{n=1}^{N} b_n b_n^T \tag{5.84}
$$

where $ b_n = \nabla y_n = \nabla a_n $, because the activation function for the output units is simply the identity. Evaluation of the outer product approximation for the Hessian is straightforward as it only involves first derivatives of the error function, which can be evaluated efficiently in $ O(W) $ steps using standard backpropagation. The elements of the matrix can then be found in $ O(W^2) $ steps by simple multiplication.

It is important to emphasize that this approximation is only likely to be valid for a network that has been trained appropriately, and that for a general network mapping, the second derivative terms on the right-hand side of (5.83) will typically not be negligible. 

In the case of the cross-entropy error function for a network with logistic sigmoid output-unit activation functions, the corresponding approximation is given by:

$$
H \approx \sum_{n=1}^{N} y_n (1 - y_n) b_n b_n^T \tag{5.85}
$$

An analogous result can be obtained for multiclass networks having softmax output-unit activation functions.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)

def mse_loss_derivative(y_pred, y_true):
    return 2 * (y_pred - y_true) / len(y_true)

# Simple 2-layer Neural Network forward and backward propagation
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases
        self.W1 = np.random.randn(self.input_size, self.hidden_size)
        self.b1 = np.zeros(self.hidden_size)
        self.W2 = np.random.randn(self.hidden_size, self.output_size)
        self.b2 = np.zeros(self.output_size)
    
    def forward(self, X):
        self.hidden_input = np.dot(X, self.W1) + self.b1
        self.hidden_output = sigmoid(self.hidden_input)
        self.output_input = np.dot(self.hidden_output, self.W2) + self.b2
        self.output = sigmoid(self.output_input)
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Compute gradients for output layer
        output_error = mse_loss_derivative(self.output, y)
        output_delta = output_error * sigmoid_derivative(self.output_input)
        
        # Compute gradients for hidden layer
        hidden_error = output_delta.dot(self.W2.T)
        hidden_delta = hidden_error * sigmoid_derivative(self.hidden_input)
        
        # Compute weight and bias gradients
        dW2 = self.hidden_output.T.dot(output_delta)
        db2 = np.sum(output_delta, axis=0)
        dW1 = X.T.dot(hidden_delta)
        db1 = np.sum(hidden_delta, axis=0)
        
        # Update weights and biases
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
        
        # Return the gradients
        return dW1, db1, dW2, db2

# Generate some toy data for training
X_train = np.random.randn(100, 2)  # 100 samples, 2 features
y_train = (np.sum(X_train, axis=1) > 0).astype(np.float32).reshape(-1, 1)  # XOR-like data

# Compute the Hessian matrix for the network
def compute_hessian(nn, X, y):
    # Perform forward pass
    y_pred = nn.forward(X)
    
    # Perform backward pass to compute gradients
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Compute second derivatives (Hessian)
    H = np.zeros((nn.input_size + nn.hidden_size + nn.output_size, nn.input_size + nn.hidden_size + nn.output_size))
    
    # Diagonal Hessian approximation
    H_diag = np.zeros_like(H)
    for i in range(nn.input_size):
        for j in range(nn.hidden_size):
            H_diag[i, j] = np.sum(dW1 ** 2)  # Approximation for diagonal
    
    # Outer product approximation
    H_outer = np.outer(dW1.flatten(), dW1.flatten())  # Using the outer product of gradients
    
    return H, H_diag, H_outer

# Create a simple neural network instance
nn = SimpleNN(input_size=2, hidden_size=4, output_size=1)

# Compute Hessian and its approximations
H, H_diag, H_outer = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Matrix and its Approximations
plt.figure(figsize=(12, 6))

# Full Hessian
plt.subplot(1, 3, 1)
plt.imshow(H, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Full Hessian")

# Diagonal Approximation
plt.subplot(1, 3, 2)
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Diagonal Approximation")

# Outer Product Approximation
plt.subplot(1, 3, 3)
plt.imshow(H_outer, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Outer Product Approximation")

plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return sum((y_pred[i] - y_true[i]) ** 2 for i in range(len(y_pred))) / len(y_pred)

def mse_loss_derivative(y_pred, y_true):
    # Compare individual scalar values in nested lists
    return [2 * (y_pred[i][0] - y_true[i][0]) / len(y_true) for i in range(len(y_pred))]

# Simple 2-layer Neural Network forward and backward propagation
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases with random values
        self.W1 = [[random.gauss(0, 1) for _ in range(self.hidden_size)] for _ in range(self.input_size)]
        self.b1 = [0] * self.hidden_size
        self.W2 = [[random.gauss(0, 1) for _ in range(self.output_size)] for _ in range(self.hidden_size)]
        self.b2 = [0] * self.output_size
    
    def forward(self, X):
        # Forward pass through hidden layer
        self.hidden_input = [
            [sum(X[k][i] * self.W1[i][j] for i in range(self.input_size)) + self.b1[j] for j in range(self.hidden_size)]
            for k in range(len(X))  # Iterate over each sample in the dataset
        ]
        self.hidden_output = [
            [sigmoid(x) for x in sample_input] for sample_input in self.hidden_input
        ]
        
        # Forward pass through output layer
        self.output_input = [
            [sum(self.hidden_output[k][i] * self.W2[i][j] for i in range(self.hidden_size)) + self.b2[j] for j in range(self.output_size)]
            for k in range(len(X))  # Iterate over each sample in the dataset
        ]
        
        self.output = [
            [sigmoid(x) for x in sample_input] for sample_input in self.output_input
        ]
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Compute gradients for output layer
        output_error = mse_loss_derivative(self.output, y)
        output_delta = [
            [output_error[k][i] * sigmoid_derivative(self.output_input[k][i]) for i in range(self.output_size)]
            for k in range(len(X))
        ]
        
        # Compute gradients for hidden layer
        hidden_error = [
            [sum(output_delta[k][j] * self.W2[i][j] for j in range(self.output_size)) for i in range(self.hidden_size)]
            for k in range(len(X))
        ]
        hidden_delta = [
            [hidden_error[k][i] * sigmoid_derivative(self.hidden_input[k][i]) for i in range(self.hidden_size)]
            for k in range(len(X))
        ]
        
        # Compute weight and bias gradients
        dW2 = [
            [sum(self.hidden_output[k][i] * output_delta[k][j] for k in range(len(X))) for i in range(self.hidden_size)]
            for j in range(self.output_size)
        ]
        db2 = [sum(output_delta[k][j] for k in range(len(X))) for j in range(self.output_size)]
        dW1 = [
            [sum(X[k][i] * hidden_delta[k][j] for k in range(len(X))) for i in range(self.input_size)]
            for j in range(self.hidden_size)
        ]
        db1 = [sum(hidden_delta[k][j] for k in range(len(X))) for j in range(self.hidden_size)]
        
        # Update weights and biases
        self.W2 = [
            [self.W2[i][j] - learning_rate * dW2[j][i] for j in range(self.output_size)]
            for i in range(self.hidden_size)
        ]
        self.b2 = [self.b2[i] - learning_rate * db2[i] for i in range(self.output_size)]
        self.W1 = [
            [self.W1[i][j] - learning_rate * dW1[j][i] for j in range(self.hidden_size)]
            for i in range(self.input_size)
        ]
        self.b1 = [self.b1[i] - learning_rate * db1[i] for i in range(self.hidden_size)]
        
        return dW1, db1, dW2, db2

# Generate some toy data for training
def generate_data(n_samples, input_size):
    X = [[random.gauss(0, 1) for _ in range(input_size)] for _ in range(n_samples)]
    y = [[1 if sum(X[i]) > 0 else 0] for i in range(n_samples)]  # XOR-like data
    return X, y

# Compute the Hessian matrix for the network
def compute_hessian(nn, X, y):
    # Perform forward pass
    y_pred = nn.forward(X)
    
    # Perform backward pass to compute gradients
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Flatten the gradients to compute the Hessian
    W1_flat = [item for sublist in nn.W1 for item in sublist]
    W2_flat = [item for sublist in nn.W2 for item in sublist]
    
    # Hessian matrix of second derivatives (approximation)
    H = [[0] * (len(W1_flat) + len(W2_flat)) for _ in range(len(W1_flat) + len(W2_flat))]
    
    # Diagonal Hessian approximation (second derivatives of loss w.r.t parameters)
    H_diag = [[0] * (len(W1_flat) + len(W2_flat)) for _ in range(len(W1_flat) + len(W2_flat))]
    for i in range(len(W1_flat)):
        H_diag[i][i] = sum(dW1[i] ** 2)  # Approximation for diagonal
    
    # Outer product approximation (for each layer)
    H_outer = [[dW1[i] * dW2[j] for j in range(len(W2_flat))] for i in range(len(W1_flat))]
    
    return H, H_diag, H_outer

# Create a simple neural network instance
input_size = 2
hidden_size = 4
output_size = 1
nn = SimpleNN(input_size, hidden_size, output_size)

# Generate training data
X_train, y_train = generate_data(100, input_size)

# Compute Hessian and its approximations
H, H_diag, H_outer = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Matrix and its Approximations
plt.figure(figsize=(12, 6))

# Full Hessian (this is just a placeholder, no computation for full Hessian in this version)
plt.subplot(1, 3, 1)
plt.imshow(H, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Full Hessian")

# Diagonal Approximation
plt.subplot(1, 3, 2)
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Diagonal Approximation")

# Outer Product Approximation
plt.subplot(1, 3, 3)
plt.imshow(H_outer, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Outer Product Approximation")

plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error Loss function and its derivative
def mse_loss(y_pred, y_true):
    return sum((y_pred[i] - y_true[i]) ** 2 for i in range(len(y_pred))) / len(y_pred)

def mse_loss_derivative(y_pred, y_true):
    return [(y_pred[i] - y_true[i]) * 2 / len(y_true) for i in range(len(y_pred))]

# Simple Neural Network
class SimpleNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases with random values
        self.W1 = [[random.gauss(0, 1) for _ in range(hidden_size)] for _ in range(input_size)]
        self.b1 = [0] * hidden_size
        self.W2 = [[random.gauss(0, 1) for _ in range(output_size)] for _ in range(hidden_size)]
        self.b2 = [0] * output_size
    
    def forward(self, X):
        # Forward pass through hidden layer
        self.hidden_input = [[sum(X[k][i] * self.W1[i][j] for i in range(self.input_size)) + self.b1[j] for j in range(self.hidden_size)] for k in range(len(X))]
        self.hidden_output = [[sigmoid(val) for val in sample] for sample in self.hidden_input]
        
        # Forward pass through output layer
        self.output_input = [[sum(self.hidden_output[k][i] * self.W2[i][j] for i in range(self.hidden_size)) + self.b2[j] for j in range(self.output_size)] for k in range(len(X))]
        self.output = [[sigmoid(val) for val in sample] for sample in self.output_input]
        return self.output
    
    def backward(self, X, y, learning_rate=0.1):
        # Backpropagation
        output_error = mse_loss_derivative(self.output, y)
        output_delta = [[output_error[k] * sigmoid_derivative(self.output_input[k][0]) for _ in range(self.output_size)] for k in range(len(X))]
        
        hidden_error = [[sum(output_delta[k][j] * self.W2[i][j] for j in range(self.output_size)) for i in range(self.hidden_size)] for k in range(len(X))]
        hidden_delta = [[hidden_error[k][i] * sigmoid_derivative(self.hidden_input[k][i]) for i in range(self.hidden_size)] for k in range(len(X))]
        
        # Gradient descent update for weights and biases
        dW2 = [[sum(self.hidden_output[k][i] * output_delta[k][j] for k in range(len(X))) for i in range(self.hidden_size)] for j in range(self.output_size)]
        db2 = [sum(output_delta[k][j] for k in range(len(X))) for j in range(self.output_size)]
        dW1 = [[sum(X[k][i] * hidden_delta[k][j] for k in range(len(X))) for i in range(self.input_size)] for j in range(self.hidden_size)]
        db1 = [sum(hidden_delta[k][j] for k in range(len(X))) for j in range(self.hidden_size)]
        
        # Update weights and biases
        self.W2 = [[self.W2[i][j] - learning_rate * dW2[j][i] for j in range(self.output_size)] for i in range(self.hidden_size)]
        self.b2 = [self.b2[i] - learning_rate * db2[i] for i in range(self.output_size)]
        self.W1 = [[self.W1[i][j] - learning_rate * dW1[j][i] for j in range(self.hidden_size)] for i in range(self.input_size)]
        self.b1 = [self.b1[i] - learning_rate * db1[i] for i in range(self.hidden_size)]
        
        return dW1, db1, dW2, db2

# Generate some toy data for training
def generate_data(n_samples, input_size):
    X = [[random.gauss(0, 1) for _ in range(input_size)] for _ in range(n_samples)]
    y = [[1 if sum(X[i]) > 0 else 0] for i in range(n_samples)]  # Simple decision boundary
    return X, y

# Simplified Hessian approximation
def compute_hessian(nn, X, y):
    y_pred = nn.forward(X)
    dW1, db1, dW2, db2 = nn.backward(X, y)
    
    # Diagonal approximation for Hessian
    H_diag = [[dW1[i][j] ** 2 + dW2[i][j] ** 2 for j in range(len(dW1[0]))] for i in range(len(dW1))]
    return H_diag

# Create a simple neural network instance
input_size = 2
hidden_size = 4
output_size = 1
nn = SimpleNN(input_size, hidden_size, output_size)

# Generate training data
X_train, y_train = generate_data(100, input_size)

# Compute Hessian approximation
H_diag = compute_hessian(nn, X_train, y_train)

# Visualizing Hessian Diagonal
plt.imshow(H_diag, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Hessian Diagonal Approximation")
plt.show()
```
