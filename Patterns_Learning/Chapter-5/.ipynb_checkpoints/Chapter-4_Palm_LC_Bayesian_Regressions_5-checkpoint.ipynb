## Bayesian Logistic Regression

We now turn to a Bayesian treatment of logistic regression. Exact Bayesian inference for logistic regression is intractable. In particular, evaluation of the posterior distribution would require normalization of the product of a prior distribution and a likelihood function that itself comprises a product of logistic sigmoid functions, one for every data point. Evaluation of the predictive distribution is similarly intractable. Here we consider the application of the **Laplace approximation** to the problem of Bayesian logistic regression (Spiegelhalter and Lauritzen, 1990; MacKay, 1992b).

###  Laplace Approximation

Recall from Section 4.4 that the **Laplace approximation** is obtained by finding the mode of the posterior distribution and then fitting a Gaussian centered at that mode. This requires evaluation of the second derivatives of the log posterior, which is equivalent to finding the **Hessian matrix**. 

Because we seek a Gaussian representation for the posterior distribution, it is natural to begin with a **Gaussian prior**, which we write in the general form:

$$
p(w) = \mathcal{N}(w | m_0, S_0)
$$

where $ m_0 $ and $ S_0 $ are fixed hyperparameters. The posterior distribution over $ w $ is given by:

$$
p(w|t) \propto p(w)p(t|w)
$$

where $ t = (t_1, \dots, t_N)^T $.

Taking the log of both sides, and substituting for the prior distribution using the Gaussian form, and for the likelihood function using the logistic model $ \sigma(w^T \phi_n) $, we obtain:

$$
\ln p(w|t) = -\frac{1}{2} (w - m_0)^T S_0^{-1} (w - m_0) - \frac{N}{2} + \sum_{n=1}^N \left[ t_n \ln y_n + (1 - t_n) \ln (1 - y_n) \right] + \text{const}
$$

where $ y_n = \sigma(w^T \phi_n) $ is the logistic sigmoid function.

To obtain a **Gaussian approximation** to the posterior distribution, we first maximize the posterior distribution to give the **MAP (maximum a posteriori)** solution $ w_{MAP} $, which defines the mean of the Gaussian. The covariance is then given by the inverse of the matrix of second derivatives of the negative log likelihood, which takes the form:

$$
S_N = - \nabla \nabla \ln p(w|t) = S_0^{-1} + \sum_{n=1}^N y_n (1 - y_n) \phi_n \phi_n^T
$$

The Gaussian approximation to the posterior distribution therefore takes the form:

$$
q(w) = \mathcal{N}(w | w_{MAP}, S_N)
$$

Having obtained a Gaussian approximation to the posterior distribution, there remains the task of marginalizing with respect to this distribution in order to make predictions.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.special import expit  # Sigmoid function

# Logistic Sigmoid function
def sigmoid(z):
    return expit(z)  # Sigmoid activation

# Define the Negative Log Posterior
def log_prior(w, m0, S0):
    """Compute the log of the Gaussian prior"""
    return -0.5 * np.dot(np.dot((w - m0).T, np.linalg.inv(S0)), (w - m0))

def log_likelihood(w, X, t):
    """Compute the log-likelihood of the logistic regression model"""
    y = sigmoid(np.dot(X, w))  # Predictions using the logistic model
    return np.sum(t * np.log(y) + (1 - t) * np.log(1 - y))

def negative_log_posterior(w, X, t, m0, S0):
    """Negative log posterior = Negative log prior + Negative log likelihood"""
    return -log_prior(w, m0, S0) - log_likelihood(w, X, t)

# Hessian Matrix (second derivatives of the negative log posterior)
def hessian(w, X, t):
    """Compute the Hessian matrix of the negative log posterior"""
    y = sigmoid(np.dot(X, w))
    R = np.diag(y * (1 - y))  # Diagonal matrix of the variance of the logistic model
    H = np.dot(X.T, np.dot(R, X))
    return H

# MAP Estimation (Maximizing the Posterior)
def fit_bayesian_logistic(X, t, m0, S0):
    """Fit the Bayesian Logistic Regression model using Laplace Approximation"""
    # Initialize weights randomly
    w_init = np.zeros(X.shape[1])

    # Use minimize function to find the MAP estimate
    result = minimize(negative_log_posterior, w_init, args=(X, t, m0, S0), method='BFGS')
    
    # MAP estimate of weights
    w_MAP = result.x
    return w_MAP

# Compute Posterior Covariance (using Hessian)
def compute_posterior_covariance(w_MAP, X, t, S0):
    """Compute the covariance of the posterior distribution using the Hessian"""
    H = hessian(w_MAP, X, t)
    SN_inv = np.linalg.inv(S0) + H  # Adding the prior covariance (S0)
    SN = np.linalg.inv(SN_inv)  # Posterior covariance is the inverse of the sum
    return SN

# Prediction Using Laplace Approximation
def predict(X, w_MAP, SN):
    """Make predictions using the Laplace Approximation"""
    y_pred = sigmoid(np.dot(X, w_MAP))  # Predictions (probabilities)
    
    # Uncertainty (using the covariance of the posterior)
    uncertainty = np.dot(X, np.dot(SN, X.T))  # Approximate uncertainty
    return y_pred, uncertainty

# Laplace Approximation to Model Evidence and BIC
def compute_laplace_approximation(w_MAP, SN, m0, S0):
    """Compute the Laplace Approximation to the log model evidence"""
    log_evidence = -0.5 * np.dot(w_MAP.T, np.dot(np.linalg.inv(SN), w_MAP)) + \
                   0.5 * np.log(np.linalg.det(SN)) + \
                   -0.5 * np.dot((w_MAP - m0).T, np.dot(np.linalg.inv(S0), (w_MAP - m0)))
    return log_evidence

def compute_bic(log_likelihood_val, N, M):
    """Compute the Bayesian Information Criterion (BIC)"""
    return -2 * log_likelihood_val + M * np.log(N)

# Example Data and Model Fitting
if __name__ == "__main__":
    # Generate synthetic data (2D) for binary classification
    np.random.seed(42)
    N = 100  # Number of data points
    X = np.random.randn(N, 2)  # Feature matrix (100x2)
    true_w = np.array([1, -2])  # True weights
    t = (sigmoid(np.dot(X, true_w)) > 0.5).astype(int)  # Labels using logistic function

    # Gaussian prior (with zero mean and identity covariance)
    m0 = np.zeros(X.shape[1])  # Prior mean
    S0 = np.eye(X.shape[1])  # Prior covariance

    # Fit Bayesian Logistic Regression model using Laplace Approximation
    w_MAP = fit_bayesian_logistic(X, t, m0, S0)

    # Compute posterior covariance
    SN = compute_posterior_covariance(w_MAP, X, t, S0)

    # Create a grid of points for plotting the decision boundary
    xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
    grid = np.c_[xx.ravel(), yy.ravel()]

    # Make predictions on the grid
    y_pred, uncertainty = predict(grid, w_MAP, SN)

    # Reshape for plotting
    y_pred = y_pred.reshape(xx.shape)

    # Uncertainty needs to be computed as variance of the prediction
    uncertainty = np.diagonal(np.dot(grid, np.dot(SN, grid.T)))  # Correct way to compute uncertainty

    # Reshape uncertainty to match the grid shape
    uncertainty = uncertainty.reshape(xx.shape)

    # Plot decision boundary
    plt.figure(figsize=(8, 6))

    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=t, palette="Set1", edgecolor='k', s=100)
    plt.contourf(xx, yy, y_pred, levels=[0, 0.5, 1], alpha=0.3, colors=['blue', 'red'])

    plt.title('Bayesian Logistic Regression with Laplace Approximation')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.colorbar()
    plt.show()

    # Compute Laplace Approximation to the log model evidence
    log_evidence = compute_laplace_approximation(w_MAP, SN, m0, S0)

    # Compute the log likelihood
    log_likelihood_val = log_likelihood(w_MAP, X, t)

    # Compute BIC
    BIC = compute_bic(log_likelihood_val, N, X.shape[1])

    print(f"Laplace Approximation to log model evidence: {log_evidence}")
    print(f"BIC: {BIC}")
```

##  Predictive Distribution

The predictive distribution for class $ C_1 $, given a new feature vector $ \phi(x) $, is obtained by marginalizing with respect to the posterior distribution $ p(w|t) $, which is itself approximated by a Gaussian distribution $ q(w) $. This results in the following expression:

$$
p(C_1 | \phi, t) = \int p(C_1 | \phi, w) p(w | t) \, dw = \int \sigma(w^T \phi) q(w) \, dw
$$

where $ \sigma(w^T \phi) $ is the logistic sigmoid function and $ q(w) $ is the Gaussian approximation of the posterior distribution.

### Step 1: Rewriting the Logistic Sigmoid

The logistic sigmoid function $ \sigma(w^T \phi) $ depends on $ w $ only through its projection onto $ \phi $. Denote this projection as $ a = w^T \phi $. We can express $ \sigma(w^T \phi) $ using the Dirac delta function $ \delta(a - w^T \phi) $:

$$
\sigma(w^T \phi) = \int \delta(a - w^T \phi) \sigma(a) \, da
$$

Thus, the predictive distribution becomes:

$$
\int \sigma(w^T \phi) q(w) \, dw = \int \sigma(a) p(a) \, da
$$

where \( p(a) \) is the marginal distribution of \( a \), given by:

$$
p(a) = \int \delta(a - w^T \phi) q(w) \, dw
$$

### Step 2: Marginal Distribution of $ a $

Since $ q(w) $ is Gaussian, and the delta function imposes a linear constraint on $ w $, the marginal distribution of $ a $ is also Gaussian. The mean and covariance of this distribution can be computed as follows:

- The **mean** of $ a $ is:

$$
\mu_a = \mathbb{E}[a] = \int a p(a) \, da = \int w^T \phi q(w) \, dw = w_{\text{MAP}}^T \phi
$$

where $ w_{\text{MAP}} $ is the MAP estimate of the weights.

- The **variance** of $ a $ is:

$$
\sigma_a^2 = \text{Var}[a] = \int a^2 p(a) \, da - \mu_a^2 = \int (w^T \phi)^2 q(w) \, dw - (w_{\text{MAP}}^T \phi)^2 = \phi^T S_N \phi
$$

where $ S_N $ is the posterior covariance matrix.

Thus, the marginal distribution of $ a $ is:

$$
a \sim \mathcal{N}(\mu_a, \sigma_a^2)
$$

### Step 3: Approximation to the Predictive Distribution

To evaluate the predictive distribution, we use the fact that the logistic sigmoid function $ \sigma(a) $ and the Gaussian distribution for $ a $ are closely related. To make the integration tractable, we approximate the logistic function $ \sigma(a) $ by the **probit function** $ \Phi(a) $, which is similar to the logistic function:

$$
\sigma(a) \approx \Phi(\lambda a)
$$

where $ \lambda^2 = \frac{\pi}{8} $. The advantage of using the probit function is that its convolution with a Gaussian distribution can be expressed analytically. Specifically, we have the following approximation:

$$
\int \sigma(a) \mathcal{N}(a|\mu, \sigma^2) \, da \approx \sigma\left( \frac{\mu}{\sqrt{\sigma^2 + \frac{\pi}{8}}} \right)
$$

This leads to the approximate predictive distribution:

$$
p(C_1 | \phi, t) = \sigma\left( \kappa(\sigma_a^2) \mu_a \right)
$$

where $ \kappa(\sigma_a^2) $ is a correction factor given by:

$$
\kappa(\sigma_a^2) = \left( 1 + \frac{\pi \sigma_a^2}{8} \right)^{-\frac{1}{2}}
$$

Thus, the predictive distribution for class $ C_1 $ is:

$$
p(C_1 | \phi, t) = \sigma\left( \left( 1 + \frac{\pi \sigma_a^2}{8} \right)^{-\frac{1}{2}} \mu_a \right)
$$

### Step 4: Decision Boundary

The decision boundary corresponding to $ p(C_1 | \phi, t) = 0.5 $ is given by $ \mu_a = 0 $. This is the same as the decision boundary obtained using the MAP estimate for $ w $, which corresponds to the point where the posterior mean of the projection $ w_{\text{MAP}}^T \phi $ equals zero.

### Summary

The predictive distribution for class $ C_1 $, given the feature vector $ \phi $ and training data $ t $, can be approximated using the logistic sigmoid function and the Gaussian posterior distribution:

$$
p(C_1 | \phi, t) = \sigma\left( \left( 1 + \frac{\pi \sigma_a^2}{8} \right)^{-\frac{1}{2}} w_{\text{MAP}}^T \phi \right)
$$

This approximation is particularly useful for making predictions and understanding the uncertainty in the model's decisions.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit as sigmoid  # Logistic sigmoid function

# Define the sigmoid function for logistic regression
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Logistic Regression model: using Laplace Approximation
class BayesianLogisticRegression:
    def __init__(self, X, y, prior_mean=0, prior_cov=1):
        self.X = X  # Input data
        self.y = y  # Target labels
        self.prior_mean = prior_mean  # Prior mean (often zero)
        self.prior_cov = prior_cov  # Prior covariance (often identity matrix)
        self.weights_map = None  # MAP estimate for weights
        self.posterior_cov = None  # Posterior covariance
        self.N, self.M = X.shape  # N: number of samples, M: number of features

    # Fit the model using MAP (Maximum A Posteriori)
    def fit(self):
        # Adding the intercept term to the design matrix
        X_augmented = np.hstack([np.ones((self.N, 1)), self.X])  # Add bias column
        
        # Initialize weights with zero (MAP estimate is where log-posterior is maximized)
        self.weights_map = np.zeros(self.M + 1)
        
        # Likelihood function gradient
        def log_likelihood(w):
            # Linear model
            z = np.dot(X_augmented, w)
            predictions = sigmoid(z)
            # Negative log-likelihood (to minimize)
            return -np.sum(self.y * np.log(predictions) + (1 - self.y) * np.log(1 - predictions))

        # Gradient of the log likelihood function
        def gradient_log_likelihood(w):
            z = np.dot(X_augmented, w)
            predictions = sigmoid(z)
            gradient = np.dot(X_augmented.T, predictions - self.y)
            return gradient

        # Hessian (second derivative of the log likelihood)
        def hessian_log_likelihood(w):
            z = np.dot(X_augmented, w)
            predictions = sigmoid(z)
            diag = predictions * (1 - predictions)  # Element-wise multiplication
            hessian = np.dot(X_augmented.T * diag, X_augmented)
            return hessian

        # Use Newton's method for optimization (to maximize log posterior)
        from scipy.optimize import minimize
        result = minimize(log_likelihood, self.weights_map, jac=gradient_log_likelihood, hess=hessian_log_likelihood)
        self.weights_map = result.x

        # Posterior covariance matrix is the inverse of the negative Hessian at the MAP
        hessian_at_map = hessian_log_likelihood(self.weights_map)
        self.posterior_cov = np.linalg.inv(hessian_at_map)

    # Predict the probability of class 1 given new data
    def predict(self, X_new):
        X_augmented = np.hstack([np.ones((X_new.shape[0], 1)), X_new])  # Add bias term
        z = np.dot(X_augmented, self.weights_map)
        return sigmoid(z)

    # Calculate the uncertainty in predictions
    def uncertainty(self, X_new):
        X_augmented = np.hstack([np.ones((X_new.shape[0], 1)), X_new])  # Add bias term
        variance = np.sum(X_augmented.dot(self.posterior_cov) * X_augmented, axis=1)
        return np.sqrt(variance)

# Example usage:

# Generating synthetic data
np.random.seed(42)
N = 100  # Number of samples
M = 2  # Number of features
X = np.random.randn(N, M)  # Random feature data
true_weights = np.array([2, -1, 0.5])  # True weights (including bias term)
y = np.random.binomial(1, sigmoid(np.dot(np.hstack([np.ones((N, 1)), X]), true_weights)))  # Generate binary labels

# Initialize Bayesian Logistic Regression model
blr = BayesianLogisticRegression(X, y)

# Fit the model
blr.fit()

# Make predictions and calculate uncertainty on new data
X_new = np.random.randn(100, M)  # New data
y_pred = blr.predict(X_new)  # Predicted probabilities
uncertainty = blr.uncertainty(X_new)  # Uncertainty in predictions

# Plotting the decision boundary and uncertainty
xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
grid_points = np.c_[xx.ravel(), yy.ravel()]
probs = blr.predict(grid_points).reshape(xx.shape)  # Predictions on the grid

plt.figure(figsize=(10, 6))

# Plot decision boundary
plt.contourf(xx, yy, probs, levels=np.linspace(0, 1, 20), cmap="Blues", alpha=0.5)
plt.colorbar(label="Predicted Probability")

# Plot training data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', s=100)
plt.title("Bayesian Logistic Regression with Laplace Approximation")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Plot uncertainty in predictions
plt.figure(figsize=(10, 6))
plt.scatter(X_new[:, 0], X_new[:, 1], c=uncertainty, cmap='viridis', edgecolors='k', s=100)
plt.colorbar(label="Uncertainty (Std. Dev.)")
plt.title("Uncertainty in Predictions")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import math
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Helper function for matrix multiplication
def matmul(A, B):
    return [[sum(a * b for a, b in zip(A_row, B_col)) for B_col in zip(*B)] for A_row in A]


# Helper function for matrix transposition
def transpose(M):
    return [list(i) for i in zip(*M)]


# Helper function for matrix inversion (using Gauss-Jordan elimination)
def matinv(A):
    n = len(A)
    M = [row[:] for row in A]  # Make a copy of the matrix
    I = [[1 if i == j else 0 for j in range(n)] for i in range(n)]  # Identity matrix
    
    for i in range(n):
        if M[i][i] == 0:
            for j in range(i + 1, n):
                if M[j][i] != 0:
                    M[i], M[j] = M[j], M[i]
                    I[i], I[j] = I[j], I[i]
                    break

        inv = 1 / M[i][i]
        for j in range(n):
            M[i][j] *= inv
            I[i][j] *= inv

        for j in range(n):
            if j != i:
                factor = M[j][i]
                for k in range(n):
                    M[j][k] -= factor * M[i][k]
                    I[j][k] -= factor * M[i][k]
    
    return I


# Logistic sigmoid function with clipping to prevent overflow
def sigmoid(x):
    # Clip the value to avoid overflow
    x = max(min(x, 500), -500)  # Clip between -500 and 500 to avoid overflow
    return 1 / (1 + math.exp(-x))


# Logistic Regression model with Laplace Approximation
class BayesianLogisticRegression:
    def __init__(self, X, y, prior_mean=0, prior_cov=1):
        self.X = X  # Input data
        self.y = y  # Target labels
        self.prior_mean = prior_mean  # Prior mean (often zero)
        self.prior_cov = prior_cov  # Prior covariance (often identity matrix)
        self.weights_map = None  # MAP estimate for weights
        self.posterior_cov = None  # Posterior covariance
        self.N = len(X)  # Number of samples
        self.M = len(X[0])  # Number of features

    # Fit the model using MAP (Maximum A Posteriori)
    def fit(self):
        # Augment X with bias term (add column of ones)
        X_augmented = [[1] + row for row in self.X]  # Add bias column
        
        # Initialize weights with zero (MAP estimate is where log-posterior is maximized)
        self.weights_map = [0] * (self.M + 1)
        
        # Likelihood function gradient
        def log_likelihood(w):
            z = [sum(x * w_i for x, w_i in zip(X_row, w)) for X_row in X_augmented]
            predictions = [sigmoid(z_i) for z_i in z]
            # Negative log-likelihood (to minimize)
            return -sum(t * math.log(y_i) + (1 - t) * math.log(1 - y_i) for t, y_i in zip(self.y, predictions))

        # Gradient of the log likelihood function
        def gradient_log_likelihood(w):
            z = [sum(x * w_i for x, w_i in zip(X_row, w)) for X_row in X_augmented]
            predictions = [sigmoid(z_i) for z_i in z]
            gradient = [sum((pred - t) * X_augmented[i][j] for i, (pred, t) in enumerate(zip(predictions, self.y))) for j in range(self.M + 1)]
            return gradient

        # Hessian (second derivative of the log likelihood)
        def hessian_log_likelihood(w):
            z = [sum(x * w_i for x, w_i in zip(X_row, w)) for X_row in X_augmented]
            predictions = [sigmoid(z_i) for z_i in z]
            diag = [pred * (1 - pred) for pred in predictions]
            hessian = [[sum(diag[i] * X_augmented[i][k] * X_augmented[i][l] for i in range(self.N)) for l in range(self.M + 1)] for k in range(self.M + 1)]
            return hessian

        # Use Newton's method for optimization (to maximize log posterior)
        def newton_method():
            w = self.weights_map
            for _ in range(100):  # Max iterations
                grad = gradient_log_likelihood(w)
                hess = hessian_log_likelihood(w)
                hess_inv = matinv(hess)
                w = [w_i - sum(hess_inv[i][j] * grad[j] for j in range(self.M + 1)) for i, w_i in enumerate(w)]
            return w
        
        self.weights_map = newton_method()

        # Posterior covariance matrix is the inverse of the negative Hessian at the MAP
        hessian_at_map = hessian_log_likelihood(self.weights_map)
        self.posterior_cov = matinv(hessian_at_map)

    # Predict the probability of class 1 given new data
    def predict(self, X_new):
        X_augmented = [[1] + row for row in X_new]  # Add bias term
        z = [sum(x * w_i for x, w_i in zip(X_row, self.weights_map)) for X_row in X_augmented]
        return [sigmoid(z_i) for z_i in z]

    # Calculate the uncertainty in predictions
    def uncertainty(self, X_new):
        X_augmented = [[1] + row for row in X_new]  # Add bias term
        variance = [sum(X_augmented[i][j] * self.posterior_cov[j][k] * X_augmented[i][k] for j in range(self.M + 1) for k in range(self.M + 1)) for i in range(len(X_new))]
        return [math.sqrt(v) for v in variance]


# Example usage:

# Generating synthetic data
random.seed(42)
N = 100  # Number of samples
M = 2  # Number of features
X = [[random.gauss(0, 1) for _ in range(M)] for _ in range(N)]  # Random feature data
true_weights = [2, -1, 0.5]  # True weights (including bias term)
y = [random.randint(0, 1) if sigmoid(sum(x * w for x, w in zip(X_row, true_weights))) > 0.5 else 0 for X_row in X]  # Generate binary labels

# Initialize Bayesian Logistic Regression model
blr = BayesianLogisticRegression(X, y)

# Fit the model
blr.fit()

# Make predictions and calculate uncertainty on new data
X_new = [[random.gauss(0, 1) for _ in range(M)] for _ in range(100)]  # New data
y_pred = blr.predict(X_new)  # Predicted probabilities
uncertainty = blr.uncertainty(X_new)  # Uncertainty in predictions

# Plotting the decision boundary and uncertainty
xx, yy = [[-3 + i * 0.06 for i in range(100)] for _ in range(2)]
grid_points = [[xx[i], yy[j]] for i in range(100) for j in range(100)]
probs = [blr.predict([point])[0] for point in grid_points]  # Predictions on the grid

plt.figure(figsize=(10, 6))

# Plot decision boundary
probs_2d = []
for i in range(100):
    probs_2d.append(probs[i * 100:(i + 1) * 100])
plt.contourf(xx, yy, probs_2d, levels=[0, 0.5, 1], cmap="Blues", alpha=0.5)
plt.colorbar(label="Predicted Probability")

# Plot training data
plt.scatter([row[0] for row in X], [row[1] for row in X], c=y, cmap='coolwarm', edgecolors='k', s=100)
plt.title("Bayesian Logistic Regression with Laplace Approximation")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Plot uncertainty in predictions
plt.figure(figsize=(10, 6))
plt.scatter([row[0] for row in X_new], [row[1] for row in X_new], c=uncertainty, cmap='viridis', edgecolors='k', s=100)
plt.colorbar(label="Uncertainty (Std. Dev.)")
plt.title("Uncertainty in Predictions")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
