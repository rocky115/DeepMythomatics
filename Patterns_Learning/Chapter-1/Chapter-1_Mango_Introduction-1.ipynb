## Pattern Recognition and Machine Learning

The problem of searching for patterns in data is fundamental and has a long history. For example, Tycho Brahe's extensive astronomical observations in the 16th century allowed Johannes Kepler to discover empirical laws of planetary motion. These discoveries later laid the foundation for classical mechanics.

Similarly, discovering regularities in atomic spectra played a crucial role in developing quantum physics in the early 20th century. The field of **pattern recognition** deals with automatic discovery of data regularities through algorithms and applies these regularities to classify data into various categories.

Consider recognizing handwritten digits. Each digit can be represented as a $28 \times 28$ pixel image, forming a vector $x$ of 784 real numbers. The goal is to create a machine learning model that maps an input vector $x$ to an output indicating the identity of the digit (0–9).

Building such a machine is challenging because of handwriting variability. A rule-based approach might lead to poor results due to the numerous exceptions. Instead, machine learning offers a better solution by using a large dataset $\{x_1, \dots, x_N\}$ called a **training set**, where each image is labeled with its correct digit. The **target vector** $t$ represents the corresponding digit. The learning algorithm produces a function $y(x)$, which takes a new image $x$ and outputs a vector $y$, corresponding to the predicted digit.

The function $y(x)$ is determined during the **training phase** using the training data. After training, the model can generalize to new examples, known as the **test set**. A key challenge is **generalization**, ensuring the model correctly categorizes new inputs that weren't part of the training set.

In most applications, input data undergoes **pre-processing** to simplify the recognition task. For digit recognition, images are typically translated and scaled to fit within a fixed-size box, reducing variability. Pre-processing might also reduce dimensionality, making computation faster and more efficient.

## Learning Paradigms

- **Supervised Learning**: Training data includes input vectors and corresponding target vectors. Examples include **classification** (e.g., digit recognition) and **regression** (predicting continuous variables, such as chemical yield).
  
- **Unsupervised Learning**: Training data consists only of input vectors without target values. The aim is to discover patterns like **clustering** (grouping similar examples) or **density estimation** (modeling data distribution).

- **Reinforcement Learning**: The algorithm learns by interacting with the environment to maximize a reward. Unlike supervised learning, there are no explicit examples of correct outputs. The system learns through trial and error, balancing **exploration** (trying new actions) and **exploitation** (using actions known to provide high rewards).

Reinforcement learning continues to be an active area of research.

## Polynomial Curve Fitting Example

We introduce a simple **regression** problem to motivate key concepts in curve fitting. Suppose we observe a real-valued input variable $x$, and we want to predict a real-valued target variable $t$. For illustration, consider an artificial dataset generated from the function $\sin(2\pi x)$ with added Gaussian noise, as shown in Appendix A.

Let the training set consist of $N$ observations of $x$ (denoted as $\mathbf{x} = (x_1, \dots, x_N)^\top$) and their corresponding target values $\mathbf{t} = (t_1, \dots, t_N)^\top$. The input data $x_n$ is spaced uniformly in the range $[0, 1]$, and the corresponding $t_n$ values are computed as:

$$
t_n = \sin(2\pi x_n) + \text{noise}
$$

The training set in Figure 1.2 comprises $N = 10$ data points, with noise added using a Gaussian distribution. This method mimics real-world datasets, where underlying patterns exist but individual observations are corrupted by noise.

Our goal is to predict a new value $\hat{t}$ for a given $\hat{x}$. In essence, we aim to discover the underlying function $\sin(2\pi x)$. However, because the observed data are noisy, there is uncertainty in predicting $\hat{t}$ for $\hat{x}$.

## Polynomial Curve Fitting

We begin by fitting the data with a polynomial of the form:

$$
y(x, \mathbf{w}) = w_0 + w_1 x + w_2 x^2 + \dots + w_M x^M = \sum_{j=0}^{M} w_j x^j
$$

where \(M\) is the order of the polynomial, and the coefficients $w_0, \dots, w_M$ are collectively represented as the vector $\mathbf{w}$.

Although the polynomial function $y(x, \mathbf{w})$ is nonlinear in $x$, it is linear in the parameters $\mathbf{w}$. Such models, which are linear in their unknown parameters, are called **linear models**.

## Error Function

The polynomial coefficients $\mathbf{w}$ are determined by minimizing an error function, which measures the difference between the model predictions $y(x_n, \mathbf{w})$ and the actual target values $t_n$. A common error function is the sum of squares of errors:

$$
E(\mathbf{w}) = \frac{1}{2} \sum_{n=1}^{N} \left[ y(x_n, \mathbf{w}) - t_n \right]^2
$$

The factor $\frac{1}{2} $ is included for convenience in later calculations. The error function is always non-negative and becomes zero only if the model perfectly fits the training data.

# Polynomial Curve Fitting - Error Function

The error function defined earlier corresponds to (one-half of) the sum of the squares of the vertical displacements of each data point from the function $y(x, \mathbf{w})$. Geometrically, this can be interpreted as the vertical green bars in Figure.3, showing how each training point deviates from the function.

To solve the curve fitting problem, we seek the vector $\mathbf{w}$ that minimizes the error function:

$$
E(\mathbf{w}) = \frac{1}{2} \sum_{n=1}^{N} \left[ y(x_n, \mathbf{w}) - t_n \right]^2
$$

Since this error function is quadratic in the coefficients $\mathbf{w}$, minimizing it gives a unique solution, denoted by $\mathbf{w}^\star$, which can be determined analytically. The corresponding polynomial is then given by $y(x, \mathbf{w}^\star)$.

## Model Selection

An important issue in polynomial curve fitting is selecting the order $M$ of the polynomial. Figure 1.4 shows the result of fitting polynomials of different orders $(M = 0$, $M = 1$, $M = 3$, and $M = 9$) to the dataset. Here are the key observations:

- **$M = 0$** and **$M = 1$** result in poor fits because they fail to capture the oscillatory nature of $\sin(2\pi x)$.
- **$M = 3$** seems to provide a good fit to the data and captures the overall trend of the function.
- **$M = 9$** perfectly fits the training data, but the curve oscillates wildly between the data points, a phenomenon known as **overfitting**.

## Overfitting and Generalization

Overfitting occurs when the model fits the training data too well but performs poorly on new, unseen data. To quantify this effect, we evaluate the model’s performance on a **test set**, which consists of 100 data points generated using the same process as the training set but with new random noise values.

For each choice of $M$, we compute the residual error for both the training set and the test set. A convenient measure of error is the **Root-Mean-Square (RMS) error**:

$$
E_{\text{RMS}} = \sqrt{\frac{2E(\mathbf{w}^\star)}{N}}
$$

where $N$ is the size of the dataset, allowing for comparison across datasets of different sizes, and the square root ensures that $E_{\text{RMS}}$ has the same units as the target variable $t$.

## Training vs. Test Set Error

Figure.5 shows the RMS error for both the training set and the test set as a function of \(M\). Key insights from the figure:

- For small $M$, the test set error is large because the polynomial is too inflexible to capture the oscillations of $\sin(2\pi x)$.
- For values of $M$ in the range $3 \leq M \leq 8$, the test set error is minimized, indicating good generalization.
- For large $M$, the test set error increases due to overfitting, as the model starts fitting noise in the training data.

Thus, model selection involves finding a balance between **underfitting** (low $M$) and **overfitting** (high $M$), typically through validation on a test set.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
N = 10  # Number of data points
x = np.linspace(0, 1, N)
t = np.sin(2 * np.pi * x) + 0.1 * np.random.randn(N)  # Target with added Gaussian noise

# Define the polynomial function
def polynomial_basis_function(x, degree):
    """Returns the design matrix for polynomial basis functions."""
    return np.vander(x, degree + 1, increasing=True)

# Define the mean square error function
def mean_square_error(y, t):
    """Calculates the mean square error."""
    return 0.5 * np.sum((y - t) ** 2)

# Fit the polynomial
def fit_polynomial(x, t, degree):
    """Fit a polynomial of a given degree to the data."""
    Phi = polynomial_basis_function(x, degree)
    w = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ t  # Solve for the coefficients
    return w

# Plot the fitted curve for different polynomial degrees
def plot_fitted_curves(x, t, max_degree=9):
    x_plot = np.linspace(0, 1, 1000)  # For plotting smooth curves
    t_plot = np.sin(2 * np.pi * x_plot)

    plt.figure(figsize=(10, 8))

    for degree in range(max_degree + 1):
        w = fit_polynomial(x, t, degree)
        Phi_plot = polynomial_basis_function(x_plot, degree)
        y_plot = Phi_plot @ w

        plt.subplot(2, 2, degree // 3 + 1)
        plt.plot(x_plot, t_plot, 'g', label='True Function')
        plt.scatter(x, t, color='b', zorder=5, label='Training Data')
        plt.plot(x_plot, y_plot, 'r', label=f'Polynomial (M={degree})')
        plt.title(f"Degree M={degree}")
        plt.legend()

    plt.tight_layout()
    plt.show()

# Plot the results for polynomials of degree 0, 1, 3, and 9
plot_fitted_curves(x, t, max_degree=9)
```

## Polynomial Curve Fitting: Mathematical Explanation

Polynomial curve fitting is a method of regression used to model the relationship between an input variable $ x $ and an output variable $ t $ by fitting a polynomial function to a set of data points. This is useful for approximating functions from noisy observations and is often used in machine learning to understand concepts like model complexity, underfitting, and overfitting.

## 1. Problem Setup:

We are given a set of $ N $ input-output pairs $ \{(x_n, t_n)\} $, where:

- $ x_n $ is the $ n $-th input data point.
- $ t_n $ is the corresponding target (or output) value, typically with some noise.

Our goal is to find a function $ y(x) $ that models the relationship between inputs $ x $ and target values $ t $. In polynomial regression, we use a polynomial of degree $ M $:

$$
y(x, \mathbf{w}) = w_0 + w_1 x + w_2 x^2 + \cdots + w_M x^M = \sum_{j=0}^{M} w_j x^j
$$

Here:

- $ M $ is the degree of the polynomial.
- $ \mathbf{w} = (w_0, w_1, \ldots, w_M) $ are the polynomial coefficients we need to determine.

## 2. Polynomial Basis Function:

The polynomial function $ y(x, \mathbf{w}) $ can be expressed using a design matrix $ \Phi $, where each row corresponds to the input $ x_n $ raised to successive powers. For instance, for $ x_n $, the row in $ \Phi $ would be:

$$
\Phi_n = \begin{bmatrix} 1 & x_n & x_n^2 & \cdots & x_n^M \end{bmatrix}
$$

Thus, for all data points $ x $, the design matrix is:

$$
\Phi = \begin{bmatrix} 
1 & x_1 & x_1^2 & \cdots & x_1^M \\ 
1 & x_2 & x_2^2 & \cdots & x_2^M \\ 
\vdots & \vdots & \vdots & \ddots & \vdots \\ 
1 & x_N & x_N^2 & \cdots & x_N^M 
\end{bmatrix}
$$

## 3. Error Function (Objective Function):

To find the best-fit polynomial, we need to minimize the error function (or cost function) that measures the discrepancy between the predicted values $ y(x_n, \mathbf{w}) $ and the actual target values $ t_n $.

A common choice is the sum of squared errors (SSE), given by:

$$
E(\mathbf{w}) = \frac{1}{2} \sum_{n=1}^{N} \left( y(x_n, \mathbf{w}) - t_n \right)^2
$$

This can be rewritten in matrix form as:

$$
E(\mathbf{w}) = \frac{1}{2} \left( \mathbf{y} - \mathbf{t} \right)^T \left( \mathbf{y} - \mathbf{t} \right)
$$

where:

- $ \mathbf{y} = \Phi \mathbf{w} $ is the vector of predicted outputs.
- $ \mathbf{t} $ is the vector of true target values.

## 4. Solution Using the Normal Equation:

The goal is to minimize the error function $ E(\mathbf{w}) $. Since this is a quadratic function in $ \mathbf{w} $, it has a unique solution, which can be found by taking the derivative of $ E(\mathbf{w}) $ with respect to $ \mathbf{w} $ and setting it to zero:

$$
\frac{\partial E(\mathbf{w})}{\partial \mathbf{w}} = \Phi^T \Phi \mathbf{w} - \Phi^T \mathbf{t} = 0
$$

Solving for $ \mathbf{w} $, we obtain the normal equation:

$$
\mathbf{w} = (\Phi^T \Phi)^{-1} \Phi^T \mathbf{t}
$$

This gives the coefficients $ \mathbf{w} $ that minimize the error function and provide the best-fit polynomial to the data.

## 5. Model Complexity and Overfitting:

The degree of the polynomial $ M $ plays a crucial role in the model's complexity. Choosing an appropriate $ M $ is essential for good generalization to new data.

- **Underfitting (High Bias)**: When $ M $ is too small (e.g., a linear model for a sinusoidal function), the model cannot capture the underlying pattern of the data, leading to large errors on both the training and test sets.

- **Overfitting (High Variance)**: When $ M $ is too large, the model fits the training data too closely, including the noise. While the training error may be very small (even zero), the model performs poorly on new data, resulting in poor generalization.

This trade-off between model complexity and generalization is illustrated by the bias-variance tradeoff.

## 6. Root-Mean-Square (RMS) Error:

To compare the performance across different model complexities (i.e., different values of $ M $), we often use the root-mean-square error (RMS), defined as:

$$
E_{\text{RMS}} = \sqrt{\frac{2E(\mathbf{w})}{N}}
$$

This allows us to evaluate the performance on both the training set and a separate test set to identify overfitting.

## 7. Example of Fitting Polynomials of Different Degrees:

When fitting polynomials of different degrees to the data, we observe the following:

- For $ M = 0 $ or $ M = 1 $, the model is too simple and cannot capture the complexity of the underlying function $ \sin(2\pi x) $.
- For $ M = 3 $, the model fits the data well and generalizes reasonably.
- For $ M = 9 $, the model overfits the training data, capturing noise and showing large oscillations.

This behavior is visually represented in graphs where both the training and test set errors are plotted for various polynomial degrees.

## Conclusion:

Polynomial curve fitting is a powerful technique for regression, but care must be taken to balance model complexity. The key challenge is finding a model that generalizes well to unseen data, which involves selecting the right polynomial degree and managing the bias-variance tradeoff.
